# Dataflow RunInference — Model Hot-Swap with Event Mode

Streaming inference on Google Cloud Dataflow with Apache Beam's `RunInference`, featuring **event-mode model hot-swap** — swap models at runtime via Pub/Sub without restarting the pipeline.

**What this demonstrates:**

1. **Custom ModelHandler** for HuggingFace sentiment models loaded directly from the Hub
2. **KeyedModelHandler** for metadata passthrough in streaming pipelines
3. **Event-mode model hot-swap** via Pub/Sub side input (`model_metadata_pcoll`)
4. **`PersistentRunInference`** — a `RunInference` subclass that prevents `EmptySideInput` reversion
5. **Continuous side input propagation** — ensures every DoFn instance permanently switches to the new model
6. **Bidirectional swap verification** — swap v1→v2 and v2→v1, confirming both stick permanently

> **Why event mode over watch mode?** Beam also offers `watch_model_pattern` for GCS-hosted models, but it only detects *new files* — to roll back, you'd have to copy an older model artifact back into the watched directory under a new filename. Event mode gives you explicit, intentional control: publish a single Pub/Sub message with any model identifier (name, path, version) and the pipeline swaps to it. Rollback is the same operation as deployment — just publish the previous model. See [Event Mode vs Watch Mode](#event-vs-watch-mode) below for details.

**Models used:**

| Version | Model | Params | Description |
|---------|-------|--------|-------------|
| v1 | `distilbert-base-uncased-finetuned-sst-2-english` | 67M | DistilBERT fine-tuned on SST-2 |
| v2 | `textattack/bert-base-uncased-SST-2` | 110M | BERT fine-tuned on SST-2 |

Both accept text input and output binary sentiment classification (positive/negative), making them directly interchangeable. The `model_id` field in each prediction tracks which model produced it.

**Pipeline Architecture:**

```
ReadFromPubSub(model_sub) → Parse → (key=0, (version, ModelMetadata)) ─┐
                                                                         ├─ Flatten → ModelReemitFn → GlobalWindow
PeriodicImpulse(1s) ──────→ (key=0, None) ─────────────────────────────┘       ↑ stores latest model in Beam state
                                                                                ↑ re-emits on each heartbeat
                                                                                       |
                                                                                       v (side input)
Pub/Sub Input --> parse --> Window(1s) --> PersistentRunInference --> Pub/Sub Output
```

## Custom ModelHandler

The `ModelHandler` wraps HuggingFace models for use with `RunInference`. Key methods:

- **`load_model()`** — Downloads and loads the model + tokenizer from HuggingFace Hub
- **`run_inference()`** — Tokenizes text, runs the model, returns `PredictionResult` with `model_id`
- **`update_model_path()`** — Called by Beam when a model update event arrives; updates the model name so the next `load_model()` call fetches the new model
- **`batch_elements_kwargs()`** — Controls internal batching (replaces explicit `BatchElements`)

**How model updates flow:** Side input emits `ModelMetadata` → `_RunInferenceDoFn` detects new `model_id` → calls `update_model_path()` → calls `load_model()` → all subsequent predictions reflect the new model.

````python
import json
import time
from datetime import datetime

import torch
import apache_beam as beam
from apache_beam import window
from apache_beam.transforms import trigger
from apache_beam.ml.inference.base import (
    ModelHandler, RunInference, PredictionResult,
    KeyedModelHandler, ModelMetadata,
)
from apache_beam.options.pipeline_options import PipelineOptions
from apache_beam.io.gcp.pubsub import ReadFromPubSub, WriteToPubSub
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Both models are SST-2 sentiment classifiers: index 0 = NEGATIVE, 1 = POSITIVE.
# Some HuggingFace models (e.g. textattack/*) don't set id2label in their config,
# defaulting to LABEL_0/LABEL_1. This mapping ensures consistent output labels.
SST2_LABELS = {0: 'NEGATIVE', 1: 'POSITIVE'}


class _LoadedModel:
    # Wrapper so Beam's shared model handle can create a weak reference.
    # Plain tuples don't support weakref; this class does.
    def __init__(self, model, tokenizer):
        self.model = model
        self.tokenizer = tokenizer


class HuggingFaceSentimentHandler(ModelHandler):
    # ModelHandler for HuggingFace sentiment classification models.
    #
    # Loads models directly from the HuggingFace Hub. Supports event-mode
    # hot-swap: update_model_path() changes the model name, then load_model()
    # downloads and loads the new model.

    def __init__(self, model_name):
        self._model_name = model_name

    def load_model(self):
        tokenizer = AutoTokenizer.from_pretrained(self._model_name)
        model = AutoModelForSequenceClassification.from_pretrained(self._model_name)
        model.eval()
        return _LoadedModel(model, tokenizer)

    def update_model_path(self, model_path):
        self._model_name = model_path

    def run_inference(self, batch, model, inference_args=None):
        model_obj, tokenizer = model.model, model.tokenizer
        inputs = tokenizer(
            list(batch), padding=True, truncation=True,
            max_length=128, return_tensors='pt',
        )
        with torch.no_grad():
            outputs = model_obj(**inputs)
        probs = torch.softmax(outputs.logits, dim=-1)

        results = []
        for i in range(len(batch)):
            label_idx = torch.argmax(probs[i]).item()
            label = SST2_LABELS.get(label_idx, str(label_idx))
            results.append(PredictionResult(
                example=batch[i],
                inference={'label': label, 'confidence': probs[i].max().item()},
                model_id=self._model_name,
            ))
        return results

    def batch_elements_kwargs(self):
        return {'min_batch_size': 1, 'max_batch_size': 32}
````

## Pipeline Functions

Simple functions to parse Pub/Sub input and format RunInference output.

````python
def parse_message(message):
    # Parse Pub/Sub bytes -> (metadata_key, text) for KeyedModelHandler.
    data = json.loads(message.decode('utf-8'))
    metadata_key = json.dumps({
        'review_id': data.get('review_id'),
        'publish_time': data.get('publish_time'),
    })
    return (metadata_key, data['text'])


def format_result(element):
    # Format (metadata_key, PredictionResult) -> JSON bytes for Pub/Sub output.
    metadata_key, prediction = element
    metadata = json.loads(metadata_key)

    result = {
        'text': prediction.example,
        'label': prediction.inference['label'],
        'confidence': round(prediction.inference['confidence'], 4),
        'model_id': prediction.model_id,
        'review_id': metadata.get('review_id'),
        'timestamp': datetime.utcnow().isoformat(),
    }
    return json.dumps(result).encode('utf-8')
````

## Build and Run Streaming Pipeline

<a id="side-input-design"></a>

### Side Input Design: Making Model Swaps Stick

RunInference's `model_metadata_pcoll` parameter accepts a PCollection of `ModelMetadata` and passes it internally as an [`AsSingleton`](https://beam.apache.org/releases/pydoc/current/apache_beam.pvalue.html#apache_beam.pvalue.AsSingleton) side input. When a new `model_id` arrives, workers download and load the new model — no pipeline restart needed.

Making this work reliably in streaming mode requires two things: preventing reversion, and keeping the side input populated so every DoFn instance sees the update.

### Preventing Reversion with `PersistentRunInference`

> **This is the single most important detail for event-mode hot-swap.**

`RunInference.expand()` creates the side input without a default value:

```python
beam.pvalue.AsSingleton(self._model_metadata_pcoll)   # no default_value
```

Whether `AsSingleton` has a `default_value` controls what happens when no fresh side input pane is available:

| `AsSingleton(pcoll)` — no default | `AsSingleton(pcoll, default_value=None)` — with default |
|---|---|
| Returns **`EmptySideInput`** when no pane is available | Returns **`None`** when no pane is available |
| `_RunInferenceDoFn` **actively reverts** to the original model | `_RunInferenceDoFn` **preserves** whatever model is currently loaded |

In Dataflow Streaming Engine with `DISCARDING` accumulation, side input panes are consumed almost immediately. Without a default value, `_RunInferenceDoFn` receives `EmptySideInput` on every bundle that doesn't coincide with a fresh pane — and reverts to the model that was serialized at pipeline construction time.

**`PersistentRunInference`** subclasses `RunInference` to inject `default_value=None` into the `AsSingleton` call. With this change, bundles without a fresh pane take a code path that preserves the currently loaded model.

### Keeping the Side Input Populated

`PersistentRunInference` ensures a DoFn instance never reverts, but each instance still needs to *read a pane* containing the new model at least once in order to switch. A single Pub/Sub message produces one element, which creates one side input pane — consumed almost immediately by the Streaming Engine. DoFn instances that aren't processing a bundle at that moment miss it and stay on the old model.

To ensure every instance eventually switches, the pipeline continuously re-emits the current model metadata into the side input. A `PeriodicImpulse` root source generates a heartbeat element every second. These heartbeats are merged with Pub/Sub model updates via `Flatten` and fed to `ModelReemitFn` — a stateful DoFn that stores the latest model in Beam state and re-emits it on each heartbeat. Every re-emission creates a fresh side input pane, giving DoFn instances another chance to read it and permanently switch.

This is the same pattern Beam's [`WatchFilePattern`](https://beam.apache.org/releases/pydoc/current/apache_beam.ml.inference.utils.html#apache_beam.ml.inference.utils.WatchFilePattern) uses internally — `PeriodicImpulse` → continuous re-emission → side input. `PeriodicImpulse` is a root source with its own resources, so it doesn't compete with Pub/Sub element processing.

`ModelReemitFn` also versions each update with a `time.time()` timestamp. Since Pub/Sub provides [at-least-once delivery](https://cloud.google.com/pubsub/docs/subscriber#at-least-once-delivery), old messages can be re-delivered after newer ones. The version check ensures only genuinely newer updates are accepted — a re-delivered old swap event is silently ignored.

**How a model switch propagates:**

1. You publish `{"model_path": "new-model", "version": <timestamp>}` to Pub/Sub
2. `ModelReemitFn` receives it, checks `version >= current`, stores it in Beam state, and emits it
3. On every subsequent heartbeat (1/s), `ModelReemitFn` re-emits the stored model
4. Each re-emission creates a fresh side input pane
5. As each `_RunInferenceDoFn` instance starts a new bundle, it reads a pane, detects the new `model_id`, downloads the model, and switches — permanently
6. Thanks to `PersistentRunInference`, instances that haven't read a pane yet keep their current model (no reversion)
7. Within a few minutes, all instances have read a pane and switched
8. The model stays on the new version until you intentionally publish again

<a id="event-vs-watch-mode"></a>

### Event Mode vs Watch Mode

Beam provides two ways to trigger model updates in `RunInference`:

| | Event Mode (`model_metadata_pcoll`) | Watch Mode (`watch_model_pattern`) |
|---|---|---|
| **Update trigger** | Pub/Sub message, API call, database event — any PCollection | New file detected in GCS directory |
| **Rollback** | Publish the previous model identifier | Copy old model artifact back under a new filename |
| **Model source** | Any location: HuggingFace Hub, GCS, S3, local, registry | Must be GCS-hosted artifacts |
| **Update latency** | Seconds (Pub/Sub delivery + model download) | Polling interval (default 360s) + model download |
| **Control** | Explicit and intentional — you decide when to swap | Implicit — any new file triggers a swap |

Watch mode is simpler for workflows that already deploy models as GCS artifacts. But for enterprise use cases that need intentional rollback, multi-source model loading, or integration with external orchestration (CI/CD, feature stores, model registries), event mode provides the explicit control you need.

Both approaches use the same `AsSingleton` side input mechanics internally, so the `PersistentRunInference` fix applies to watch mode too if you need to prevent reversion between polling intervals.

````python
from apache_beam.transforms.userstate import ReadModifyWriteStateSpec
from apache_beam.transforms.periodicsequence import PeriodicImpulse

HEARTBEAT_INTERVAL_SEC = 1  # PeriodicImpulse fires every 1s for fast propagation


class ModelReemitFn(beam.DoFn):
    """Stateful DoFn that stores the latest ModelMetadata and re-emits it
    on periodic heartbeats for continuous side input propagation.

    Receives two types of elements (merged via Flatten):
    - (0, (version, ModelMetadata)) from Pub/Sub: store if version >= current, emit
    - (0, None) from PeriodicImpulse heartbeat: re-emit stored model

    Uses monotonic versioning to ignore re-delivered Pub/Sub messages.
    Uses PeriodicImpulse (root source) instead of a timer to avoid
    contention with Pub/Sub element processing on the same key partition.
    """

    LATEST = ReadModifyWriteStateSpec('latest_model', beam.coders.PickleCoder())
    VERSION = ReadModifyWriteStateSpec('version', beam.coders.PickleCoder())

    def process(self, element,
                latest=beam.DoFn.StateParam(LATEST),
                version=beam.DoFn.StateParam(VERSION)):
        _, payload = element
        if payload is not None:
            # Model update from Pub/Sub — accept only if newer
            msg_version, model_metadata = payload
            current_version = version.read()
            if current_version is None or msg_version >= current_version:
                version.write(msg_version)
                latest.write(model_metadata)
                yield model_metadata
        else:
            # Heartbeat from PeriodicImpulse — re-emit stored model
            current = latest.read()
            if current is not None:
                yield current


print(f'ModelReemitFn defined (heartbeat every {HEARTBEAT_INTERVAL_SEC}s, monotonic versioning)')
````

````python
class PersistentRunInference(RunInference):
    """RunInference subclass that prevents model reversion between side input panes.

    The base RunInference uses AsSingleton(pcoll) without a default_value.
    When no fresh pane is available, AsSingleton returns EmptySideInput,
    and _RunInferenceDoFn.process() actively reverts to the original model.

    This subclass injects default_value=None, so process() receives None
    instead of EmptySideInput — taking a code path that preserves whatever
    model is currently loaded.
    """

    def expand(self, pcoll):
        original_as_singleton = beam.pvalue.AsSingleton
        def as_singleton_with_default(pcoll_inner, **kwargs):
            kwargs.setdefault('default_value', None)
            return original_as_singleton(pcoll_inner, **kwargs)
        beam.pvalue.AsSingleton = as_singleton_with_default
        try:
            return super().expand(pcoll)
        finally:
            beam.pvalue.AsSingleton = original_as_singleton


print('PersistentRunInference defined (AsSingleton default_value=None)')
````

````python
options = PipelineOptions([
    f'--project={PROJECT_ID}',
    f'--region={REGION}',
    '--runner=DataflowRunner',
    f'--temp_location=gs://{PROJECT_ID}/dataflow/temp',
    f'--staging_location=gs://{PROJECT_ID}/dataflow/staging',
    f'--requirements_file={requirements_file}',
    f'--job_name={RESOURCE_PREFIX}-{datetime.now().strftime("%Y%m%d-%H%M%S")}',
    '--streaming',
    '--save_main_session',
    '--machine_type=n1-standard-4',
    '--num_workers=1',
    '--max_num_workers=3',
    '--experiments=enable_streaming_engine',
    '--experiments=use_runner_v2',
])

p = beam.Pipeline(options=options)

# --- Model update side input ---
# Two independent sources merged via Flatten:
# 1. Pub/Sub model swap events (rare, intentional)
# 2. PeriodicImpulse heartbeats every 1s (re-emits stored model for propagation)
# ModelReemitFn stores updates in Beam state, re-emits on heartbeats,
# and uses monotonic versioning to ignore re-delivered old messages.

pubsub_models = (
    p
    | 'Read Model Updates' >> ReadFromPubSub(subscription=MODEL_UPDATE_SUB)
    | 'Parse Model Update' >> beam.Map(lambda msg: (
        0,  # fixed key for state partitioning
        (lambda d: (d['version'], ModelMetadata(
            model_id=d['model_path'],
            model_name=d['model_path'].split('/')[-1],
        )))(json.loads(msg.decode('utf-8')))
    ))
)

heartbeats = (
    p
    | 'Heartbeat' >> PeriodicImpulse(fire_interval=HEARTBEAT_INTERVAL_SEC, apply_windowing=False)
    | 'Key Heartbeat' >> beam.Map(lambda _: (0, None))
)

model_updates = (
    (pubsub_models, heartbeats)
    | 'Merge Model Sources' >> beam.Flatten()
    | 'Reemit Latest Model' >> beam.ParDo(ModelReemitFn())
    | 'Side Input Window' >> beam.WindowInto(
        window.GlobalWindows(),
        trigger=trigger.Repeatedly(trigger.AfterCount(1)),
        accumulation_mode=trigger.AccumulationMode.DISCARDING,
    )
)

# --- Main data pipeline ---
# PersistentRunInference prevents EmptySideInput from reverting the model.
# Combined with the heartbeat, this ensures:
# - Model only changes when you publish to the model update topic
# - Model only reverts when you intentionally publish the original model
# - All DoFn instances permanently switch as they read fresh panes
results = (
    p
    | 'Read from Pub/Sub' >> ReadFromPubSub(subscription=INPUT_SUB)
    | 'Parse' >> beam.Map(parse_message)
    | 'Window' >> beam.WindowInto(
        window.FixedWindows(1),
        trigger=trigger.Repeatedly(trigger.AfterProcessingTime(1)),
        accumulation_mode=trigger.AccumulationMode.DISCARDING,
    )
    | 'RunInference' >> PersistentRunInference(
        model_handler=keyed_handler,
        model_metadata_pcoll=model_updates,
    )
    | 'Format' >> beam.Map(format_result)
)

_ = results | 'Write to Pub/Sub' >> WriteToPubSub(topic=OUTPUT_TOPIC)

job = p.run()
print(f'Streaming job started!')
print(f'Monitor: https://console.cloud.google.com/dataflow/jobs/{REGION}?project={PROJECT_ID}')

# Seed the side input with the initial model. AsSingleton blocks until
# a value arrives, so the main pipeline won't process data until this
# initial model event is received.
init_event = json.dumps({'model_path': MODEL_V1, 'version': time.time()})
publisher.publish(MODEL_UPDATE_TOPIC, init_event.encode('utf-8'))
print(f'\nSeeded side input with initial model: {MODEL_V1}')
````

## Send Test Data and Verify Results

The helper function below polls the output subscription for results instead of relying on fixed sleep times. This handles variable Dataflow worker startup and processing latency.

````python
def pull_results(subscription, expected, timeout_minutes=10, poll_seconds=15):
    """Poll a Pub/Sub subscription until expected messages arrive or timeout."""
    subscriber = pubsub_v1.SubscriberClient()
    collected = []
    deadline = time.time() + timeout_minutes * 60
    print(f'Polling for {expected} result(s) (timeout: {timeout_minutes} min)...')

    while len(collected) < expected and time.time() < deadline:
        try:
            response = subscriber.pull(
                request={'subscription': subscription, 'max_messages': expected},
                timeout=poll_seconds,
            )
        except Exception:
            response = None

        if response and response.received_messages:
            for msg in response.received_messages:
                collected.append(json.loads(msg.message.data.decode('utf-8')))
            ack_ids = [m.ack_id for m in response.received_messages]
            subscriber.acknowledge(request={'subscription': subscription, 'ack_ids': ack_ids})
            print(f'  ... received {len(collected)}/{expected}')
        else:
            remaining = int(deadline - time.time())
            print(f'  ... waiting ({remaining}s remaining, {len(collected)}/{expected} so far)')

    return collected


def print_results(results, label='Results'):
    """Print formatted prediction results."""
    print(f'\n{label} ({len(results)} predictions):\n')
    for r in results:
        model_short = r['model_id'].split('/')[-1] if '/' in r['model_id'] else r['model_id']
        print(f"  {r['label']:>8} ({r['confidence']:.3f}) | {model_short}")
        print(f'           "{r["text"][:60]}"')

print('Helper functions defined.')
````

````python
SAMPLE_REVIEWS = [
    "This product is absolutely wonderful! Best purchase I've ever made.",
    "Terrible quality. Broke after one day. Complete waste of money.",
    "It's okay, nothing special. Does what it says.",
    "I love this! The quality exceeded my expectations.",
    "Awful experience. Customer service was unhelpful and rude.",
    "Pretty good for the price. Would recommend to others.",
    "Do not buy this. Worst product I've ever used.",
    "Amazing! Fast shipping and great quality. Five stars!",
    "Mediocre at best. There are better options out there.",
    "Fantastic product! Works perfectly and arrived on time.",
]

publisher = pubsub_v1.PublisherClient()
for i, review in enumerate(SAMPLE_REVIEWS):
    message = json.dumps({
        'text': review,
        'review_id': f'review-{i:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    })
    publisher.publish(INPUT_TOPIC, message.encode('utf-8'))

print(f'Sent {len(SAMPLE_REVIEWS)} reviews. Waiting for v1 predictions...')

# Poll until all results arrive (handles worker startup + processing latency)
v1_results = pull_results(OUTPUT_SUB, expected=len(SAMPLE_REVIEWS))
print_results(v1_results, label='Model v1 (DistilBERT)')
````

### Swap and Verify

Publish a model swap event, allow a brief propagation window for model download and side input refresh, then verify all predictions use the new model. This provides a clean proof that `PersistentRunInference` prevents reversion — once the model loads, it sticks.

````python
# Swap v1 → v2, then verify
print('Publishing swap: v1 → v2...')
swap_event = json.dumps({'model_path': MODEL_V2, 'version': time.time()})
publisher.publish(MODEL_UPDATE_TOPIC, swap_event.encode('utf-8'))

print('Waiting 3 minutes for model download and propagation...')
time.sleep(180)

print('Sending 10 reviews...')
for i, review in enumerate(SAMPLE_REVIEWS):
    message = json.dumps({
        'text': review,
        'review_id': f'verify-{i:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    })
    publisher.publish(INPUT_TOPIC, message.encode('utf-8'))

results = pull_results(OUTPUT_SUB, expected=10, timeout_minutes=5)
print_results(results, 'After v1 → v2 swap')

v2 = sum(1 for r in results if r['model_id'] == MODEL_V2)
print(f'\n{"✓" if v2 == 10 else "✗"} {v2}/10 predictions on v2 — swap {"verified" if v2 == 10 else "incomplete"}')
````

### Live Transition Monitor

Observe the model swap in real-time while data flows continuously.

**Transition behavior:** Side inputs in Dataflow Streaming Engine refresh periodically, not on every bundle boundary. During active processing, some DoFn instances pick up the new model before others, creating a brief mixed-model period. Thanks to `PersistentRunInference`, each instance that catches the update keeps it permanently — the transition is monotonic (instances only move forward to the new model, never back).

**Production pipelines:** In production, you don't pause your input — and you don't need to. The Streaming Engine refreshes side inputs on its own schedule regardless of data flow. The heartbeat ensures fresh panes are always available when it does. The swap is **eventually consistent**: all DoFn instances converge to the new model, typically within a few minutes. Multi-worker pipelines converge faster because each worker refreshes independently. Monitor the `model_id` field in your output to confirm convergence.

In this experiment, we pause data briefly after the observation window to let propagation complete on our single-worker setup, then verify the model sticks.

````python
# --- Live Transition Monitor: v2 → v1 ---

# Phase 1: Confirm v2 is active
print('Phase 1: Confirming v2 predictions...')
for i in range(5):
    msg = json.dumps({
        'text': SAMPLE_REVIEWS[i], 'review_id': f'pre-{i:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    })
    publisher.publish(INPUT_TOPIC, msg.encode('utf-8'))

pre = pull_results(OUTPUT_SUB, expected=5, timeout_minutes=2)
v2_pre = sum(1 for r in pre if r['model_id'] == MODEL_V2)
print(f'  {v2_pre}/5 v2 predictions confirmed.\n')

# Phase 2: Publish swap and observe transition (~60s)
# This shows how the Streaming Engine handles the transition during active
# processing. Some DoFn instances may pick up the new model while others
# haven't refreshed their side input yet.
print('Phase 2: Publishing swap v2 → v1, observing for 60 seconds...')
publisher.publish(MODEL_UPDATE_TOPIC,
    json.dumps({'model_path': MODEL_V1, 'version': time.time()}).encode('utf-8'))
print('  Swap event published.\n')

v1_count, v2_count, msg_num = 0, 0, 0
end_time = time.time() + 60

while time.time() < end_time:
    msg_num += 1
    review = SAMPLE_REVIEWS[msg_num % len(SAMPLE_REVIEWS)]
    publisher.publish(INPUT_TOPIC, json.dumps({
        'text': review, 'review_id': f'live-{msg_num:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    }).encode('utf-8'))
    time.sleep(2)

    try:
        response = subscriber.pull(
            request={'subscription': OUTPUT_SUB, 'max_messages': 10}, timeout=3)
    except Exception:
        continue
    if not response or not response.received_messages:
        continue

    for m in response.received_messages:
        data = json.loads(m.message.data.decode('utf-8'))
        model = data['model_id'].split('/')[-1] if '/' in data['model_id'] else data['model_id']
        is_v1 = data['model_id'] == MODEL_V1
        if is_v1:
            v1_count += 1
        else:
            v2_count += 1
        marker = '  <-- v1!' if is_v1 and v1_count == 1 else ''
        print(f'  {model:<30} {data["label"]:<10} "{data["text"][:40]}"{marker}')
    subscriber.acknowledge(request={'subscription': OUTPUT_SUB,
        'ack_ids': [m.ack_id for m in response.received_messages]})

print(f'\n  Transition snapshot: {v2_count} v2, {v1_count} v1')

# Phase 3: Propagation window
# Allow the Streaming Engine to refresh the side input across all DoFn
# instances. In production, this happens naturally — no pause needed.
# Here we pause to demonstrate that the model sticks once propagated.
print(f'\nPhase 3: Waiting 3 minutes for full propagation...')
time.sleep(180)

# Phase 4: Verify model sticks
print('\nPhase 4: Sending final batch to verify model sticks...')
for i, review in enumerate(SAMPLE_REVIEWS):
    publisher.publish(INPUT_TOPIC, json.dumps({
        'text': review, 'review_id': f'final-{i:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    }).encode('utf-8'))

final = pull_results(OUTPUT_SUB, expected=10, timeout_minutes=2)
print_results(final, 'After propagation window')

v1_final = sum(1 for r in final if r['model_id'] == MODEL_V1)
if v1_final == 10:
    print(f'\n✓ {v1_final}/10 on v1 — model swap is permanent. No reversion.')
else:
    print(f'\n  {v1_final}/10 v1, {10 - v1_final}/10 v2 — propagation still in progress.')
````

## Summary

This notebook demonstrated event-mode model hot-swap on Dataflow using Apache Beam's `RunInference` with Pub/Sub-triggered model updates. Two design concerns were addressed:

| Concern | Solution | Detail |
|---------|----------|--------|
| **Reversion** — `EmptySideInput` reverts model between panes | `PersistentRunInference` | Injects `AsSingleton(default_value=None)` so bundles without a fresh pane preserve the current model |
| **Propagation** — each DoFn instance must read a pane to switch | Continuous re-emission via `ModelReemitFn` | Stores latest model in Beam state, re-emits every 1s via `PeriodicImpulse`, versions updates to ignore Pub/Sub re-delivery |

**Key takeaways:**

- **`AsSingleton` `default_value` is critical** — without it, `EmptySideInput` actively reverts the model between panes. This applies to both event mode and watch mode
- **The side input must stay populated** — a single Pub/Sub pane expires almost immediately. Continuously re-emitting the stored model ensures every DoFn instance eventually reads a pane and permanently switches
- **Model swap is eventually consistent** — the Streaming Engine refreshes side inputs on its own schedule. All DoFn instances converge to the new model within a few minutes. In production with multiple workers, monitor `model_id` in your output to confirm convergence
- **Event mode gives you rollback** — publish any model identifier to swap; rollback is the same operation as deployment. Watch mode requires copying old artifacts back under new filenames
- **Seed the side input** — `AsSingleton` blocks until a value arrives; publish the initial model path right after `p.run()`
