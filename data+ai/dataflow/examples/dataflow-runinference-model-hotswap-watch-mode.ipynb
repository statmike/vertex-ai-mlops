# Dataflow RunInference — Model Hot-Swap with Watch Mode

Streaming inference on Google Cloud Dataflow with Apache Beam's `RunInference`, featuring **watch-mode model hot-swap** — swap models at runtime by uploading new artifacts to GCS, no pipeline restart needed.

**What this demonstrates:**

1. **Custom ModelHandler** for HuggingFace sentiment models stored as GCS artifacts
2. **Watch-mode model hot-swap** via `watch_model_pattern` — Beam polls GCS for new model files
3. **`PersistentRunInference`** — prevents `EmptySideInput` reversion between polling intervals
4. **Swap and rollback** — upload v2 to swap, re-upload v1 under a new name to roll back

Watch mode uses [`WatchFilePattern`](https://beam.apache.org/releases/pydoc/current/apache_beam.ml.inference.utils.html#apache_beam.ml.inference.utils.WatchFilePattern) internally, which polls a GCS directory with `MatchContinuously` on a configurable interval. When a new file matching the pattern appears, it triggers a model update.

> **Need explicit rollback control?** Watch mode requires uploading model artifacts to GCS under a new filename to trigger a swap — including rollbacks. For workflows that need intentional, event-driven control (Pub/Sub, CI/CD triggers, model registries), see the [event-mode notebook](dataflow-runinference-model-hotswap-event-mode.ipynb) which provides rollback as a simple publish operation.

**Models used:**

| Version | Model | Params | Description |
|---------|-------|--------|-------------|
| v1 | `distilbert-base-uncased-finetuned-sst-2-english` | 67M | DistilBERT fine-tuned on SST-2 |
| v2 | `textattack/bert-base-uncased-SST-2` | 110M | BERT fine-tuned on SST-2 |

Both are saved as full artifacts to GCS using `save_pretrained()`. The pipeline loads them from GCS, not from HuggingFace Hub.

**Pipeline Architecture:**

```
WatchFilePattern(gs://bucket/.../model-*/config.json, interval=30s)
       |
       v (side input)
Pub/Sub Input --> parse --> Window(1s) --> PersistentRunInference --> Pub/Sub Output
```

````python
import tempfile
import subprocess
from transformers import AutoModelForSequenceClassification, AutoTokenizer

def save_model_to_gcs(hub_name, gcs_dir):
    """Download a model from HuggingFace Hub and upload to GCS."""
    with tempfile.TemporaryDirectory() as tmpdir:
        print(f'  Downloading {hub_name} from Hub...')
        model = AutoModelForSequenceClassification.from_pretrained(hub_name)
        tokenizer = AutoTokenizer.from_pretrained(hub_name)
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)
        print(f'  Uploading to {gcs_dir}/ ...')
        subprocess.run(
            ['gsutil', '-m', '-q', 'cp', '-r', f'{tmpdir}/*', f'{gcs_dir}/'],
            check=True, capture_output=True,
        )
    print(f'  Done: {gcs_dir}/')

# Clean any previous runs
subprocess.run(['gsutil', '-m', '-q', 'rm', '-r', f'{GCS_MODEL_DIR}/'],
               capture_output=True)  # ignore errors if dir doesn't exist

# Upload v1 as model-001 (must exist before pipeline starts)
print('Saving v1 to GCS (initial model):')
save_model_to_gcs(MODEL_V1, f'{GCS_MODEL_DIR}/model-001')

print(f'\nv1 ready at {GCS_MODEL_DIR}/model-001/')
print('v2 will be uploaded later to trigger a swap.')
````

## Custom ModelHandler

Similar to the event-mode handler, but `load_model()` downloads from a GCS path instead of loading from HuggingFace Hub. When `WatchFilePattern` detects a new file, it calls `update_model_path()` with the matched file path (e.g., `gs://bucket/.../model-002/config.json`). The handler extracts the parent directory and downloads the model artifacts.

**How model updates flow:** `WatchFilePattern` detects new `config.json` → `_RunInferenceDoFn` calls `update_model_path(file_path)` → `load_model()` downloads from GCS directory → all subsequent predictions reflect the new model.

**Handling pre-existing files:** `WatchFilePattern` internally compares file timestamps against the pipeline start time. Files uploaded *before* the pipeline started (like our initial model-001) are considered "already seen" and `update_model_path` receives an empty string. The handler guards against this — `if model_path:` — so it keeps the initial model loaded from the constructor path.

````python
import json
import os
import tempfile
import time
from datetime import datetime

import torch
import apache_beam as beam
from apache_beam import window
from apache_beam.transforms import trigger
from apache_beam.ml.inference.base import (
    ModelHandler, RunInference, PredictionResult,
    KeyedModelHandler, ModelMetadata,
)
from apache_beam.options.pipeline_options import PipelineOptions
from apache_beam.io.gcp.pubsub import ReadFromPubSub, WriteToPubSub
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from google.cloud import storage

SST2_LABELS = {0: 'NEGATIVE', 1: 'POSITIVE'}


class _LoadedModel:
    def __init__(self, model, tokenizer):
        self.model = model
        self.tokenizer = tokenizer


def _download_gcs_dir(gcs_dir, local_dir):
    """Download all files from a GCS directory to a local directory."""
    path = gcs_dir.replace('gs://', '')
    bucket_name, prefix = path.split('/', 1)
    prefix = prefix.rstrip('/')

    client = storage.Client()
    blobs = list(client.bucket(bucket_name).list_blobs(prefix=prefix + '/'))
    for blob in blobs:
        rel_path = blob.name[len(prefix) + 1:]
        if rel_path:
            local_path = os.path.join(local_dir, rel_path)
            os.makedirs(os.path.dirname(local_path), exist_ok=True)
            blob.download_to_filename(local_path)


class HuggingFaceSentimentHandler(ModelHandler):

    def __init__(self, model_path):
        self._model_path = model_path  # GCS directory path

    def load_model(self):
        local_dir = tempfile.mkdtemp()
        _download_gcs_dir(self._model_path, local_dir)
        tokenizer = AutoTokenizer.from_pretrained(local_dir)
        model = AutoModelForSequenceClassification.from_pretrained(local_dir)
        model.eval()
        return _LoadedModel(model, tokenizer)

    def update_model_path(self, model_path):
        # WatchFilePattern emits model_id='' for files older than pipeline start.
        # Ignore empty paths to keep the current model loaded.
        if model_path:
            self._model_path = os.path.dirname(model_path)

    def run_inference(self, batch, model, inference_args=None):
        model_obj, tokenizer = model.model, model.tokenizer
        inputs = tokenizer(
            list(batch), padding=True, truncation=True,
            max_length=128, return_tensors='pt',
        )
        with torch.no_grad():
            outputs = model_obj(**inputs)
        probs = torch.softmax(outputs.logits, dim=-1)

        results = []
        for i in range(len(batch)):
            label_idx = torch.argmax(probs[i]).item()
            label = SST2_LABELS.get(label_idx, str(label_idx))
            results.append(PredictionResult(
                example=batch[i],
                inference={'label': label, 'confidence': probs[i].max().item()},
                model_id=self._model_path,
            ))
        return results

    def batch_elements_kwargs(self):
        return {'min_batch_size': 1, 'max_batch_size': 32}
````

````python
initial_model_path = f'{GCS_MODEL_DIR}/model-001'
handler = HuggingFaceSentimentHandler(initial_model_path)
keyed_handler = KeyedModelHandler(handler)

print(f'Handler created for: {initial_model_path}')
print(f'Wrapped with KeyedModelHandler for metadata passthrough')
````

## Pipeline Functions

Simple functions to parse Pub/Sub input and format RunInference output.

````python
def parse_message(message):
    data = json.loads(message.decode('utf-8'))
    metadata_key = json.dumps({
        'review_id': data.get('review_id'),
        'publish_time': data.get('publish_time'),
    })
    return (metadata_key, data['text'])


def format_result(element):
    metadata_key, prediction = element
    metadata = json.loads(metadata_key)
    # Extract model directory name (e.g., 'model-001') from GCS path
    model_dir = prediction.model_id.rstrip('/').split('/')[-1] if prediction.model_id else 'unknown'

    result = {
        'text': prediction.example,
        'label': prediction.inference['label'],
        'confidence': round(prediction.inference['confidence'], 4),
        'model_id': prediction.model_id,
        'model_dir': model_dir,
        'review_id': metadata.get('review_id'),
        'timestamp': datetime.utcnow().isoformat(),
    }
    return json.dumps(result).encode('utf-8')
````

## Build and Run Streaming Pipeline

### Watch Mode Side Input

The `watch_model_pattern` parameter tells `RunInference` to poll GCS for files matching a glob pattern. Internally, it creates a [`WatchFilePattern`](https://beam.apache.org/releases/pydoc/current/apache_beam.ml.inference.utils.html#apache_beam.ml.inference.utils.WatchFilePattern) PCollection that uses `MatchContinuously` with a `PeriodicImpulse` to check for new files on the configured interval. When a new file appears with a timestamp newer than the last seen, it emits a `ModelMetadata` that triggers a model reload.

### Preventing Reversion with `PersistentRunInference`

`PersistentRunInference` is needed for watch mode for the same reason as event mode. `RunInference.expand()` creates the side input without a default value:

```python
beam.pvalue.AsSingleton(self._model_metadata_pcoll)   # no default_value
```

Whether `AsSingleton` has a `default_value` controls what happens when no fresh side input pane is available:

| `AsSingleton(pcoll)` — no default | `AsSingleton(pcoll, default_value=None)` — with default |
|---|---|
| Returns **`EmptySideInput`** when no pane is available | Returns **`None`** when no pane is available |
| `_RunInferenceDoFn` **actively reverts** to the original model | `_RunInferenceDoFn` **preserves** whatever model is currently loaded |

`PersistentRunInference` injects `default_value=None` so bundles between polling intervals preserve whatever model is currently loaded. This is the same fix as the [event-mode notebook](dataflow-runinference-model-hotswap-event-mode.ipynb) — the underlying `AsSingleton` mechanics are identical.

### Simpler Than Event Mode

Watch mode handles the continuous polling internally — no `ModelReemitFn`, no heartbeat, no monotonic versioning. `WatchFilePattern` uses the same `PeriodicImpulse` → continuous re-emission pattern that event mode builds manually, but it's all encapsulated. The pipeline is just:

```python
PersistentRunInference(
    model_handler=keyed_handler,
    watch_model_pattern='gs://bucket/.../model-*/config.json',
    interval=30,  # poll every 30s
)
```

````python
class PersistentRunInference(RunInference):
    """RunInference subclass that prevents model reversion between side input panes.

    The base RunInference uses AsSingleton(pcoll) without a default_value.
    When no fresh pane is available, AsSingleton returns EmptySideInput,
    and _RunInferenceDoFn.process() actively reverts to the original model.

    This subclass injects default_value=None, so process() receives None
    instead of EmptySideInput — taking a code path that preserves whatever
    model is currently loaded.
    """

    def expand(self, pcoll):
        original_as_singleton = beam.pvalue.AsSingleton
        def as_singleton_with_default(pcoll_inner, **kwargs):
            kwargs.setdefault('default_value', None)
            return original_as_singleton(pcoll_inner, **kwargs)
        beam.pvalue.AsSingleton = as_singleton_with_default
        try:
            return super().expand(pcoll)
        finally:
            beam.pvalue.AsSingleton = original_as_singleton


print('PersistentRunInference defined (AsSingleton default_value=None)')
````

````python
import tempfile

requirements_content = 'torch\ntransformers\ngoogle-cloud-storage\n'
with tempfile.NamedTemporaryFile(mode='w', suffix='.txt', delete=False) as f:
    f.write(requirements_content)
    requirements_file = f.name

print(f'Worker requirements written to: {requirements_file}')
````

````python
options = PipelineOptions([
    f'--project={PROJECT_ID}',
    f'--region={REGION}',
    '--runner=DataflowRunner',
    f'--temp_location=gs://{PROJECT_ID}/dataflow/temp',
    f'--staging_location=gs://{PROJECT_ID}/dataflow/staging',
    f'--requirements_file={requirements_file}',
    f'--job_name={RESOURCE_PREFIX}-{datetime.now().strftime("%Y%m%d-%H%M%S")}',
    '--streaming',
    '--save_main_session',
    '--machine_type=n1-standard-4',
    '--num_workers=1',
    '--max_num_workers=3',
    '--experiments=enable_streaming_engine',
    '--experiments=use_runner_v2',
])

p = beam.Pipeline(options=options)

# --- Main data pipeline ---
# watch_model_pattern polls GCS every WATCH_INTERVAL seconds.
# PersistentRunInference prevents EmptySideInput from reverting the model
# between polling intervals.
results = (
    p
    | 'Read from Pub/Sub' >> ReadFromPubSub(subscription=INPUT_SUB)
    | 'Parse' >> beam.Map(parse_message)
    | 'Window' >> beam.WindowInto(
        window.FixedWindows(1),
        trigger=trigger.Repeatedly(trigger.AfterProcessingTime(1)),
        accumulation_mode=trigger.AccumulationMode.DISCARDING,
    )
    | 'RunInference' >> PersistentRunInference(
        model_handler=keyed_handler,
        watch_model_pattern=WATCH_PATTERN,
        interval=WATCH_INTERVAL,
    )
    | 'Format' >> beam.Map(format_result)
)

_ = results | 'Write to Pub/Sub' >> WriteToPubSub(topic=OUTPUT_TOPIC)

job = p.run()
print(f'Streaming job started!')
print(f'Monitor: https://console.cloud.google.com/dataflow/jobs/{REGION}?project={PROJECT_ID}')
print(f'\nWatching: {WATCH_PATTERN}')
print(f'Polling every {WATCH_INTERVAL}s for new model files.')
````

## Send Test Data and Verify Results

The helper functions below wait for the Dataflow pipeline to be fully operational before sending data. Workers need time to start, install dependencies, and load the initial model from GCS — the warmup probe loop confirms end-to-end readiness by sending messages until a result comes back. The drain function clears any stale messages between test phases to keep results clean.

````python
def wait_for_job_running(job, poll_seconds=15, timeout_minutes=12):
    """Wait for the Dataflow job to reach RUNNING state."""
    from apache_beam.runners.runner import PipelineState
    deadline = time.time() + timeout_minutes * 60
    terminal = {PipelineState.DONE, PipelineState.FAILED, PipelineState.CANCELLED}
    print(f'Waiting for job to reach RUNNING state (timeout: {timeout_minutes} min)...')
    while time.time() < deadline:
        state = job.state
        if state == PipelineState.RUNNING:
            print(f'Job is RUNNING!')
            return True
        if state in terminal:
            print(f'Job reached terminal state: {state}')
            return False
        remaining = int(deadline - time.time())
        print(f'  ... {state} ({remaining}s remaining)')
        time.sleep(poll_seconds)
    print('Timeout waiting for job to start.')
    return False


def wait_for_pipeline_ready(job, input_topic, output_sub, timeout_minutes=20):
    """Wait for the pipeline to be fully operational end-to-end.

    RUNNING just means the Dataflow job is active. Workers still need to:
    install requirements, download the model, and load it into memory.
    This sends probe messages until a result comes back, confirming readiness.
    """
    wait_for_job_running(job)

    print('Warming up workers (installing deps + loading model from GCS)...')
    pub = pubsub_v1.PublisherClient()
    sub = pubsub_v1.SubscriberClient()
    deadline = time.time() + timeout_minutes * 60

    while time.time() < deadline:
        # Send a probe message
        probe = json.dumps({'text': 'warmup probe', 'review_id': 'warmup', 'publish_time': ''})
        pub.publish(input_topic, probe.encode('utf-8'))
        try:
            response = sub.pull(
                request={'subscription': output_sub, 'max_messages': 100},
                timeout=30,
            )
            if response.received_messages:
                ack_ids = [m.ack_id for m in response.received_messages]
                sub.acknowledge(request={'subscription': output_sub, 'ack_ids': ack_ids})
                print(f'Pipeline is ready! (drained {len(ack_ids)} warmup result(s))')
                # Drain any remaining warmup results
                drain_subscription(output_sub)
                return True
        except Exception:
            pass
        remaining = int(deadline - time.time())
        print(f'  ... workers not ready yet ({remaining}s remaining)')

    print('Timeout waiting for pipeline readiness.')
    return False


def drain_subscription(subscription, timeout=10):
    """Pull and discard any pending messages in a subscription."""
    subscriber = pubsub_v1.SubscriberClient()
    drained = 0
    while True:
        try:
            response = subscriber.pull(
                request={'subscription': subscription, 'max_messages': 100},
                timeout=timeout,
            )
        except Exception:
            break
        if not response.received_messages:
            break
        ack_ids = [m.ack_id for m in response.received_messages]
        subscriber.acknowledge(request={'subscription': subscription, 'ack_ids': ack_ids})
        drained += len(ack_ids)
    if drained:
        print(f'  Drained {drained} stale message(s) from output subscription.')


def pull_results(subscription, expected, timeout_minutes=10, poll_seconds=15):
    """Poll a Pub/Sub subscription until expected messages arrive or timeout."""
    subscriber = pubsub_v1.SubscriberClient()
    collected = []
    deadline = time.time() + timeout_minutes * 60
    print(f'Polling for {expected} result(s) (timeout: {timeout_minutes} min)...')

    while len(collected) < expected and time.time() < deadline:
        try:
            response = subscriber.pull(
                request={'subscription': subscription, 'max_messages': expected},
                timeout=poll_seconds,
            )
        except Exception:
            response = None

        if response and response.received_messages:
            for msg in response.received_messages:
                collected.append(json.loads(msg.message.data.decode('utf-8')))
            ack_ids = [m.ack_id for m in response.received_messages]
            subscriber.acknowledge(request={'subscription': subscription, 'ack_ids': ack_ids})
            print(f'  ... received {len(collected)}/{expected}')
        else:
            remaining = int(deadline - time.time())
            print(f'  ... waiting ({remaining}s remaining, {len(collected)}/{expected} so far)')

    return collected


def print_results(results, label='Results'):
    """Print formatted prediction results."""
    print(f'\n{label} ({len(results)} predictions):\n')
    for r in results:
        model_dir = r.get('model_dir', r['model_id'].rstrip('/').split('/')[-1])
        print(f"  {r['label']:>8} ({r['confidence']:.3f}) | {model_dir}")
        print(f'           "{r["text"][:60]}"')

print('Helper functions defined.')
````

````python
SAMPLE_REVIEWS = [
    "This product is absolutely wonderful! Best purchase I've ever made.",
    "Terrible quality. Broke after one day. Complete waste of money.",
    "It's okay, nothing special. Does what it says.",
    "I love this! The quality exceeded my expectations.",
    "Awful experience. Customer service was unhelpful and rude.",
    "Pretty good for the price. Would recommend to others.",
    "Do not buy this. Worst product I've ever used.",
    "Amazing! Fast shipping and great quality. Five stars!",
    "Mediocre at best. There are better options out there.",
    "Fantastic product! Works perfectly and arrived on time.",
]

# Wait for workers to be fully operational (RUNNING + deps installed + model loaded)
wait_for_pipeline_ready(job, INPUT_TOPIC, OUTPUT_SUB)

publisher = pubsub_v1.PublisherClient()
for i, review in enumerate(SAMPLE_REVIEWS):
    message = json.dumps({
        'text': review,
        'review_id': f'review-{i:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    })
    publisher.publish(INPUT_TOPIC, message.encode('utf-8'))

print(f'Sent {len(SAMPLE_REVIEWS)} reviews. Waiting for v1 predictions...')

v1_results = pull_results(OUTPUT_SUB, expected=len(SAMPLE_REVIEWS))
print_results(v1_results, label='Model v1 (model-001 / DistilBERT)')
````

### Swap: Upload v2 to GCS

Upload the v2 model (BERT) as `model-002/`. `WatchFilePattern` will detect the new `config.json` on its next polling cycle and trigger a model reload.

**Propagation timing:** After uploading, we wait for the polling interval (30s) plus time for the model to download on the worker and for the Streaming Engine to refresh the side input. The swap is **eventually consistent** — all DoFn instances converge to the new model, typically within a few minutes. Monitor the `model_id` field in your output to confirm convergence.

````python
print('Saving v2 to GCS to trigger swap...')
save_model_to_gcs(MODEL_V2, f'{GCS_MODEL_DIR}/model-002')

# Wait for: polling interval + model download + propagation
wait_time = WATCH_INTERVAL + 180
print(f'\nWaiting {wait_time}s for polling cycle + model download + propagation...')
time.sleep(wait_time)

# Drain any stale messages before sending fresh data
drain_subscription(OUTPUT_SUB)

print('Sending 10 reviews...')
for i, review in enumerate(SAMPLE_REVIEWS):
    message = json.dumps({
        'text': review,
        'review_id': f'swap-{i:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    })
    publisher.publish(INPUT_TOPIC, message.encode('utf-8'))

results = pull_results(OUTPUT_SUB, expected=10, timeout_minutes=5)
print_results(results, 'After v1 \u2192 v2 swap')

v2 = sum(1 for r in results if 'model-002' in r.get('model_id', ''))
print(f'\n{"\u2713" if v2 == 10 else "\u2717"} {v2}/10 predictions on model-002 (v2) \u2014 swap {"verified" if v2 == 10 else "incomplete"}')
````

### Rollback: Re-upload v1 Under a New Name

To roll back in watch mode, upload the previous model again under a **new directory name**. `WatchFilePattern` only detects files it hasn't seen before — re-using the same path (`model-001`) would be ignored.

This is the key limitation of watch mode: rollback requires re-uploading model artifacts under a new name. In [event mode](dataflow-runinference-model-hotswap-event-mode.ipynb), rollback is a single Pub/Sub publish with the previous model identifier.

````python
print('Rollback: saving v1 again as model-003...')
save_model_to_gcs(MODEL_V1, f'{GCS_MODEL_DIR}/model-003')

wait_time = WATCH_INTERVAL + 180
print(f'\nWaiting {wait_time}s for polling cycle + model download + propagation...')
time.sleep(wait_time)

# Drain any stale messages before sending fresh data
drain_subscription(OUTPUT_SUB)

print('Sending 10 reviews...')
for i, review in enumerate(SAMPLE_REVIEWS):
    message = json.dumps({
        'text': review,
        'review_id': f'rollback-{i:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    })
    publisher.publish(INPUT_TOPIC, message.encode('utf-8'))

results = pull_results(OUTPUT_SUB, expected=10, timeout_minutes=5)
print_results(results, 'After rollback to v1 (model-003)')

v1 = sum(1 for r in results if 'model-003' in r.get('model_id', ''))
print(f'\n{"\u2713" if v1 == 10 else "\u2717"} {v1}/10 predictions on model-003 (v1) \u2014 rollback {"verified" if v1 == 10 else "incomplete"}')
````

## Cleanup

**Important:** Streaming Dataflow jobs run continuously until cancelled, incurring ongoing charges. Run the cell below to stop the job and delete all resources created by this notebook.

````python
# Cancel the Dataflow streaming job
print('Cancelling Dataflow job...')
try:
    job.cancel()
    print('Job cancellation requested.')
except Exception as e:
    print(f'Could not cancel job: {e}')

# Delete Pub/Sub subscriptions
print('\nDeleting Pub/Sub subscriptions...')
subscriber = pubsub_v1.SubscriberClient()
for sub in [INPUT_SUB, OUTPUT_SUB]:
    try:
        subscriber.delete_subscription(request={'subscription': sub})
        print(f'  Deleted: {sub}')
    except Exception as e:
        print(f'  Could not delete {sub}: {e}')

# Delete Pub/Sub topics
print('\nDeleting Pub/Sub topics...')
publisher = pubsub_v1.PublisherClient()
for topic in [INPUT_TOPIC, OUTPUT_TOPIC]:
    try:
        publisher.delete_topic(request={'topic': topic})
        print(f'  Deleted: {topic}')
    except Exception as e:
        print(f'  Could not delete {topic}: {e}')

# Delete GCS model artifacts
print('\nDeleting GCS model artifacts...')
result = subprocess.run(['gsutil', '-m', '-q', 'rm', '-r', f'{GCS_MODEL_DIR}/'],
                       capture_output=True, text=True)
if result.returncode == 0:
    print(f'  Deleted: {GCS_MODEL_DIR}/')
else:
    print(f'  Could not delete {GCS_MODEL_DIR}/: {result.stderr}')

print('\nCleanup complete!')
````

## Summary

This notebook demonstrated watch-mode model hot-swap on Dataflow using Apache Beam's `RunInference` with GCS-triggered model updates. Three design concerns were addressed:

| Concern | Solution | Detail |
|---------|----------|--------|
| **Reversion** — `EmptySideInput` reverts model between polling intervals | `PersistentRunInference` | Injects `AsSingleton(default_value=None)` so bundles without a fresh pane preserve the current model |
| **Pre-existing files** — initial model uploaded before pipeline start has an older timestamp | `if model_path:` guard in `update_model_path` | `WatchFilePattern` emits empty `model_id` for files older than pipeline start time; the guard keeps the initial model loaded |
| **Rollback** — watch mode only detects *new* files | Re-upload under a new directory name | Upload the previous model as `model-003/` to trigger a fresh detection |

**Watch Mode vs Event Mode:**

| | Watch Mode | Event Mode |
|---|---|---|
| **Trigger** | New file in GCS directory | Pub/Sub message |
| **Rollback** | Re-upload old model under a new filename | Publish previous model identifier |
| **Pipeline code** | `watch_model_pattern` — no custom DoFns | `model_metadata_pcoll` + `ModelReemitFn` + heartbeat |
| **`PersistentRunInference`** | Required | Required |
| **Model source** | GCS artifacts | Any (Hub, GCS, registry, etc.) |

**Key takeaways:**

- **`AsSingleton` `default_value` is critical** — without it, `EmptySideInput` actively reverts the model between panes. This applies to both event mode and watch mode
- **Watch mode is simpler** — no heartbeat, no stateful DoFn, no versioning. `WatchFilePattern` encapsulates the same `PeriodicImpulse` → continuous re-emission pattern that event mode builds manually
- **Guard against empty model paths** — `WatchFilePattern` emits `model_id=''` for files that predate the pipeline. Your `ModelHandler` must handle this gracefully or `load_model()` will fail
- **Model swap is eventually consistent** — the Streaming Engine refreshes side inputs on its own schedule. All DoFn instances converge to the new model within a few minutes. In production with multiple workers, monitor `model_id` in your output to confirm convergence
- **Rollback is the tradeoff** — watch mode only detects new files, so rolling back means re-uploading the old model under a new name. If you need explicit control, event-driven triggers, or instant rollback, use [event mode](dataflow-runinference-model-hotswap-event-mode.ipynb)
