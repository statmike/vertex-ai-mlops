# Dataflow RunInference — Basics

Streaming ML inference on Google Cloud Dataflow using Apache Beam's `RunInference` API. This notebook teaches the fundamentals: what RunInference is, how to build a custom `ModelHandler`, and how to create a simple streaming pipeline that reads from Pub/Sub, applies a sentiment model, and writes results to Pub/Sub.

**What this demonstrates:**

1. Building a custom `ModelHandler` for HuggingFace models
2. Wrapping it with `KeyedModelHandler` for metadata passthrough
3. A simple streaming pipeline: Pub/Sub → RunInference → Pub/Sub
4. Deploying to Dataflow and verifying results

**What this does NOT cover** (covered in the hot-swap notebooks):

- Model hot-swap at runtime (see event-mode and watch-mode notebooks)
- `PersistentRunInference` and `EmptySideInput` handling
- Side inputs and stateful DoFns

**Model used:**

| Model | Params | Description |
|-------|--------|-------------|
| `distilbert-base-uncased-finetuned-sst-2-english` | 67M | DistilBERT fine-tuned on SST-2 |

Accepts text input and outputs binary sentiment classification (positive/negative). The `model_id` field in each prediction tracks which model produced it.

**Pipeline Architecture:**

```
Pub/Sub Input → Parse → Window(1s) → RunInference → Format → Pub/Sub Output
```

No side inputs, no model updates, no `PersistentRunInference` — just vanilla `RunInference`.

## Custom ModelHandler — Step by Step

The `ModelHandler` is the interface between your model and `RunInference`. It tells Beam how to load, run, and batch your model. You implement three methods:

- **`load_model()`** — Called once per worker when the pipeline starts. Downloads and returns the model object. Expensive operations (downloading weights, compiling graphs) go here so they happen once, not per-element. Must return an object that supports weak references (Beam's shared model cache uses `weakref.ref()` internally), so we wrap the model and tokenizer in a `_LoadedModel` class instead of returning a plain tuple.
- **`run_inference(batch, model, inference_args)`** — Processes a batch of elements using the model returned by `load_model()`. Returns a list of `PredictionResult` objects. Beam handles batching automatically — you just process whatever batch you receive.
- **`batch_elements_kwargs()`** — Controls how Beam groups elements into batches before calling `run_inference()`. Set `min_batch_size` and `max_batch_size` to tune throughput vs. latency.

> **Note:** This handler is intentionally simple — it does NOT implement `update_model_path()` for runtime model swapping. That's covered in the [event-mode](dataflow-runinference-model-hotswap-event-mode.ipynb) and [watch-mode](dataflow-runinference-model-hotswap-watch-mode.ipynb) notebooks.

````python
import json
import torch
from datetime import datetime

import apache_beam as beam
from apache_beam import window
from apache_beam.transforms import trigger
from apache_beam.ml.inference.base import (
    ModelHandler, RunInference, PredictionResult, KeyedModelHandler,
)
from apache_beam.options.pipeline_options import PipelineOptions
from apache_beam.io.gcp.pubsub import ReadFromPubSub, WriteToPubSub
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Both SST-2 models use index 0 = NEGATIVE, 1 = POSITIVE.
# Some HuggingFace models don't set id2label in their config,
# defaulting to LABEL_0/LABEL_1. This mapping ensures consistent output labels.
SST2_LABELS = {0: 'NEGATIVE', 1: 'POSITIVE'}


class _LoadedModel:
    # Wrapper so Beam's shared model handle can create a weak reference.
    # Plain tuples don't support weakref; this class does.
    def __init__(self, model, tokenizer):
        self.model = model
        self.tokenizer = tokenizer


class HuggingFaceSentimentHandler(ModelHandler):
    """ModelHandler for HuggingFace sentiment classification.

    This is a minimal handler — no model hot-swap support.
    For hot-swap, see the event-mode and watch-mode notebooks.
    """

    def __init__(self, model_name):
        self._model_name = model_name

    def load_model(self):
        """Download and load the model. Called once per worker."""
        tokenizer = AutoTokenizer.from_pretrained(self._model_name)
        model = AutoModelForSequenceClassification.from_pretrained(self._model_name)
        model.eval()
        return _LoadedModel(model, tokenizer)

    def run_inference(self, batch, model, inference_args=None):
        """Run inference on a batch of text strings."""
        inputs = model.tokenizer(
            list(batch), padding=True, truncation=True,
            max_length=128, return_tensors='pt',
        )
        with torch.no_grad():
            outputs = model.model(**inputs)
        probs = torch.softmax(outputs.logits, dim=-1)

        results = []
        for i in range(len(batch)):
            label_idx = torch.argmax(probs[i]).item()
            results.append(PredictionResult(
                example=batch[i],
                inference={
                    'label': SST2_LABELS.get(label_idx, str(label_idx)),
                    'confidence': probs[i].max().item(),
                },
                model_id=self._model_name,
            ))
        return results

    def batch_elements_kwargs(self):
        """Control batching: process 1-32 elements per inference call."""
        return {'min_batch_size': 1, 'max_batch_size': 32}
````

## KeyedModelHandler

`KeyedModelHandler` wraps a `ModelHandler` to pass through a key alongside each element:

- **Input:** `(key, element)` — the key carries metadata (review_id, timestamp)
- **Output:** `(key, PredictionResult)` — the key passes through inference unchanged

This is essential in streaming pipelines where you need to correlate predictions with source metadata. Without it, you'd lose track of which input produced which prediction.

````python
handler = HuggingFaceSentimentHandler(MODEL_NAME)
keyed_handler = KeyedModelHandler(handler)

print(f'Handler: {MODEL_NAME}')
print('Wrapped with KeyedModelHandler for metadata passthrough')
````

---
## Pipeline Functions

Simple functions to parse Pub/Sub input and format RunInference output.

````python
def parse_message(message):
    """Parse Pub/Sub bytes -> (metadata_key, text) for KeyedModelHandler."""
    data = json.loads(message.decode('utf-8'))
    metadata_key = json.dumps({
        'review_id': data.get('review_id'),
        'publish_time': data.get('publish_time'),
    })
    return (metadata_key, data['text'])


def format_result(element):
    """Format (metadata_key, PredictionResult) -> JSON bytes for output."""
    metadata_key, prediction = element
    metadata = json.loads(metadata_key)
    result = {
        'text': prediction.example,
        'label': prediction.inference['label'],
        'confidence': round(prediction.inference['confidence'], 4),
        'model_id': prediction.model_id,
        'review_id': metadata.get('review_id'),
        'timestamp': datetime.utcnow().isoformat(),
    }
    return json.dumps(result).encode('utf-8')
````

## Build and Run Streaming Pipeline

The pipeline reads JSON messages from Pub/Sub, extracts text and metadata, runs sentiment classification via `RunInference`, and writes predictions back to Pub/Sub.

```
Pub/Sub Input → Parse → Window(1s) → RunInference → Format → Pub/Sub Output
```

No side inputs, no model updates, no `PersistentRunInference` — just vanilla `RunInference`. The model is loaded once when workers start and stays the same for the lifetime of the pipeline.

````python
import tempfile

requirements_content = 'torch\ntransformers\n'
with tempfile.NamedTemporaryFile(mode='w', suffix='.txt', delete=False) as f:
    f.write(requirements_content)
    requirements_file = f.name

print(f'Worker requirements written to: {requirements_file}')
````

````python
options = PipelineOptions([
    f'--project={PROJECT_ID}',
    f'--region={REGION}',
    '--runner=DataflowRunner',
    f'--temp_location=gs://{PROJECT_ID}/dataflow/temp',
    f'--staging_location=gs://{PROJECT_ID}/dataflow/staging',
    f'--requirements_file={requirements_file}',
    f'--job_name={RESOURCE_PREFIX}-{datetime.now().strftime("%Y%m%d-%H%M%S")}',
    '--streaming',
    '--save_main_session',
    '--machine_type=n1-standard-4',
    '--num_workers=1',
    '--max_num_workers=2',
    '--experiments=enable_streaming_engine',
    '--experiments=use_runner_v2',
])

p = beam.Pipeline(options=options)

results = (
    p
    | 'Read from Pub/Sub' >> ReadFromPubSub(subscription=INPUT_SUB)
    | 'Parse' >> beam.Map(parse_message)
    | 'Window' >> beam.WindowInto(
        window.FixedWindows(1),
        trigger=trigger.Repeatedly(trigger.AfterProcessingTime(1)),
        accumulation_mode=trigger.AccumulationMode.DISCARDING,
    )
    | 'RunInference' >> RunInference(model_handler=keyed_handler)
    | 'Format' >> beam.Map(format_result)
)

_ = results | 'Write to Pub/Sub' >> WriteToPubSub(topic=OUTPUT_TOPIC)

job = p.run()
print(f'Streaming pipeline started!')
print(f'Monitor: https://console.cloud.google.com/dataflow/jobs/{REGION}?project={PROJECT_ID}')
````

## Send Test Data and Verify Results

The helper function below polls the output subscription for results instead of relying on fixed sleep times. This handles variable Dataflow worker startup and processing latency.

````python
import time

def pull_results(subscription, expected, timeout_minutes=10, poll_seconds=15):
    """Poll subscription until expected messages arrive or timeout."""
    collected = []
    deadline = time.time() + timeout_minutes * 60
    print(f'Polling for {expected} result(s) (timeout: {timeout_minutes} min)...')
    while len(collected) < expected and time.time() < deadline:
        try:
            response = subscriber.pull(
                request={'subscription': subscription, 'max_messages': expected},
                timeout=poll_seconds,
            )
        except Exception:
            response = None
        if response and response.received_messages:
            for msg in response.received_messages:
                collected.append(json.loads(msg.message.data.decode('utf-8')))
            ack_ids = [m.ack_id for m in response.received_messages]
            subscriber.acknowledge(request={'subscription': subscription, 'ack_ids': ack_ids})
            print(f'  ... received {len(collected)}/{expected}')
        else:
            remaining = int(deadline - time.time())
            print(f'  ... waiting ({remaining}s remaining, {len(collected)}/{expected} so far)')
    return collected
````

````python
SAMPLE_REVIEWS = [
    "This product is absolutely wonderful! Best purchase I've ever made.",
    "Terrible quality. Broke after one day. Complete waste of money.",
    "It's okay, nothing special. Does what it says.",
    "I love this! The quality exceeded my expectations.",
    "Awful experience. Customer service was unhelpful and rude.",
    "Pretty good for the price. Would recommend to others.",
    "Do not buy this. Worst product I've ever used.",
    "Amazing! Fast shipping and great quality. Five stars!",
    "Mediocre at best. There are better options out there.",
    "Fantastic product! Works perfectly and arrived on time.",
]

for i, review in enumerate(SAMPLE_REVIEWS):
    message = json.dumps({
        'text': review,
        'review_id': f'review-{i:03d}',
        'publish_time': datetime.now(datetime.now().astimezone().tzinfo).isoformat(),
    })
    publisher.publish(INPUT_TOPIC, message.encode('utf-8'))

print(f'Sent {len(SAMPLE_REVIEWS)} reviews. Waiting for predictions...')
results = pull_results(OUTPUT_SUB, expected=len(SAMPLE_REVIEWS))

print(f'\nResults ({len(results)} predictions):')
for r in results:
    model_short = r['model_id'].split('/')[-1] if '/' in r['model_id'] else r['model_id']
    print(f"  {r['label']:>8} ({r['confidence']:.3f}) | {model_short}")
    print(f'           "{r["text"][:60]}"')
````

## Summary

This notebook demonstrated the fundamentals of Apache Beam's `RunInference` API on Dataflow — building a custom `ModelHandler`, wrapping it with `KeyedModelHandler`, and deploying a simple streaming pipeline.

**Key takeaways:**

- `ModelHandler` is the interface between your model and `RunInference`
- `load_model()` runs once per worker — expensive operations go here
- `run_inference()` processes batches — keep it efficient
- `KeyedModelHandler` passes metadata through inference unchanged
- `PredictionResult` bundles the input, prediction, and `model_id`

**RunInference vs explicit DoFn approach:**

| Aspect | RunInference | Explicit DoFn |
|--------|-------------|---------------|
| Model lifecycle | Managed by Beam | You manage it |
| Batching | Built-in `BatchElements` | Manual |
| Model tracking | `PredictionResult.model_id` | Custom |
| Model swapping | Built-in via side inputs | Custom |
| Setup complexity | ModelHandler interface | More code |
