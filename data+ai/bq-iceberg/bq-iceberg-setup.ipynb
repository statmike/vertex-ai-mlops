# BigQuery Iceberg — Setup

Create an **Apache Iceberg managed table** in BigQuery from a **large** public dataset, so it shards into **many Parquet files** — then the companion R notebook can demonstrate genuinely *multithreaded*, parallel reads across those files.

- **[Apache Iceberg managed table](https://docs.cloud.google.com/bigquery/docs/biglake-iceberg-tables-in-bigquery)** (formerly *BigLake tables for Apache Iceberg in BigQuery*) — fully managed by BigQuery, but the bytes are ordinary **Parquet** in *your own* GCS bucket (`/data` shards + a `/metadata` Iceberg snapshot). Because the files are open Parquet, R `arrow` (and Spark, etc.) can read them directly and in parallel.

This notebook owns the **one-time setup** and is a **prerequisite** for: [Framework Workflows/R/R - Reading BigQuery Iceberg Tables](../../Framework%20Workflows/R/R%20-%20Reading%20BigQuery%20Iceberg%20Tables.ipynb).

**The Data:** [`bigquery-public-data.chicago_taxi_trips.taxi_trips`](https://console.cloud.google.com/marketplace/product/city-of-chicago-public-data/chicago-taxi-trips) — ~212M rows / ~83 GB (the same large table the [bq-reads](../tabular-data/bq-reads/readme.md) benchmark uses). Big enough that the Iceberg table is written as **many** Parquet shards, which is what makes the R multithreading demo real.

**Two table types, one source:**
- **Native** reads use the public taxi table **directly** — it is already a native BigQuery table, so there is no need to copy 83 GB into your project.
- **Iceberg managed** table is created here from that source (this *does* write ~83 GB of Parquet to your bucket).

> ⚠️ **Cost & runtime.** Creating the Iceberg table scans ~83 GB (~$0.40 one-time at on-demand rates) and writes ~83 GB of Parquet to your bucket (ongoing storage until you run **Cleanup**). The `CREATE TABLE` takes a few minutes. To work smaller/cheaper, set `ROW_LIMIT` below to a subset (e.g. 10_000_000).

## Setup

```python
from google.cloud import bigquery

# resolve the default project from the environment via the BigQuery client
# (more robust than parsing `gcloud config get-value`, whose output can include
# notice lines like "Your active configuration is: [...]").
bq = bigquery.Client(location='us')
PROJECT_ID = bq.project
PROJECT_ID
```

```python
REGION = 'us'                       # BigQuery + connection location (multi-region)
DATASET = 'r'                       # dataset that holds the Iceberg table
GCS_BUCKET = PROJECT_ID             # bucket that holds the Iceberg table's Parquet data

# Source: large public table (native BigQuery table) — read directly for native patterns.
SOURCE = 'bigquery-public-data.chicago_taxi_trips.taxi_trips'   # ~212M rows / ~83 GB

# Set to an int (e.g. 10_000_000) to create a smaller/cheaper Iceberg table; None = full table.
ROW_LIMIT = None

ICEBERG_TABLE = 'taxi_iceberg'                                  # Apache Iceberg managed table
STORAGE_URI   = f'gs://{GCS_BUCKET}/{DATASET}/{ICEBERG_TABLE}/iceberg'

CONNECTION_ID = 'biglake-iceberg'
CONNECTION_NAME = f'{PROJECT_ID}.{REGION}.{CONNECTION_ID}'      # used in DDL: WITH CONNECTION

# the names the R notebook reads:
NATIVE_FQN  = SOURCE                                            # public taxi table, read directly
ICEBERG_FQN = f'{PROJECT_ID}.{DATASET}.{ICEBERG_TABLE}'
print('native (read directly):', NATIVE_FQN)
print('iceberg (created here):', ICEBERG_FQN)
print('storage_uri           :', STORAGE_URI)
```

```python
# re-create the client pinned to REGION for all subsequent calls
bq = bigquery.Client(project=PROJECT_ID, location=REGION)
print("BigQuery client:", PROJECT_ID, "/", REGION)
```

## 1. BigLake Cloud Resource Connection

An Apache Iceberg managed table reaches its GCS bucket through a **Cloud Resource connection** whose service account must be granted access to the bucket.

```python
from google.cloud import bigquery_connection_v1 as bqconn

conn_client = bqconn.ConnectionServiceClient()
parent = f"projects/{PROJECT_ID}/locations/{REGION}"
conn_path = f"{parent}/connections/{CONNECTION_ID}"

try:
    conn = conn_client.get_connection(name=conn_path)
    print(f"connection exists: {CONNECTION_NAME}")
except Exception:
    conn = conn_client.create_connection(
        parent=parent,
        connection_id=CONNECTION_ID,
        connection=bqconn.Connection(cloud_resource=bqconn.CloudResourceProperties()),
    )
    print(f"created connection: {CONNECTION_NAME}")

conn_sa = conn.cloud_resource.service_account_id
print("connection service account:", conn_sa)
```

Grant the connection's service account read/write on the bucket (with retry — a freshly created connection SA can take a moment to become visible to IAM):

```python
import subprocess, time

def grant(role, attempts=8):
    for i in range(attempts):
        result = subprocess.run(
            ["gcloud", "storage", "buckets", "add-iam-policy-binding", f"gs://{GCS_BUCKET}",
             f"--member=serviceAccount:{conn_sa}", f"--role={role}"],
            capture_output=True, text=True,
        )
        if result.returncode == 0:
            return
        if "does not exist" in result.stderr and i < attempts - 1:
            wait = 5 * (i + 1)
            print(f"  SA not visible yet; retrying {role} in {wait}s...")
            time.sleep(wait)
            continue
        raise RuntimeError(result.stderr.strip())

for role in ("roles/storage.objectUser", "roles/storage.legacyBucketReader"):
    grant(role)
print("granted objectUser + legacyBucketReader to", conn_sa)
```

## 4. Apache Iceberg Managed Table

`CREATE TABLE ... WITH CONNECTION ... OPTIONS(file_format='PARQUET', table_format='ICEBERG', storage_uri=...)` creates a **managed** Iceberg table: BigQuery owns it (DML, compaction, metadata) but writes the data as **Parquet** under `storage_uri`. At ~83 GB the data is written as **many** Parquet shards — exactly what enables parallel reads in R.

> This is the expensive step (scans + writes ~83 GB). Set `ROW_LIMIT` above to create a smaller table.

```python
limit_clause = f"LIMIT {ROW_LIMIT}" if ROW_LIMIT else ""
iceberg_sql = f'''
CREATE OR REPLACE TABLE `{ICEBERG_FQN}`
WITH CONNECTION `{CONNECTION_NAME}`
OPTIONS (
    file_format = 'PARQUET',
    table_format = 'ICEBERG',
    storage_uri = '{STORAGE_URI}'
)
AS (
    SELECT * FROM `{SOURCE}` {limit_clause}
)
'''
print(iceberg_sql)
job = bq.query(iceberg_sql)
job.result()   # waits; scans ~83 GB if ROW_LIMIT is None
print("created:", ICEBERG_FQN)
```

```python
n = list(bq.query(f"SELECT COUNT(*) AS n FROM `{ICEBERG_FQN}`").result())[0]["n"]
print(f"{ICEBERG_FQN}: {n:,} rows  (Parquet under {STORAGE_URI})")
```

**Confirm the Iceberg table sharded into many Parquet files** — this is what makes the R multithreaded read meaningful:

```python
from google.cloud import storage
gcs = storage.Client(project=PROJECT_ID)
prefix = STORAGE_URI.replace(f"gs://{GCS_BUCKET}/", "") + "/data"
blobs = [b for b in gcs.list_blobs(GCS_BUCKET, prefix=prefix) if b.name.endswith(".parquet")]
total_gb = sum(b.size for b in blobs) / 1e9
print(f"{len(blobs)} Parquet data file(s), {total_gb:.1f} GB total, under {STORAGE_URI}/data")
for b in blobs[:5]:
    print(f"  {b.size/1e6:8.1f} MB  {b.name.split('/')[-1]}")
if len(blobs) > 5:
    print(f"  ... and {len(blobs)-5} more")
```

## 5. Read From Python

BigQuery-client read paths — both work on the native *and* Iceberg managed table, since BigQuery presents them through the same query/Storage interface. These reads are **bounded** (aggregate / limited) — the full table is far too large to pull into memory.

### BigFrames — pandas-like API over BigQuery

`bigframes` defers work to BigQuery; only results come back. `.shape` here is metadata, not a full download.

```python
import bigframes.pandas as bpd
bpd.options.bigquery.project = PROJECT_ID
bpd.options.bigquery.location = REGION

bdf = bpd.read_gbq_table(ICEBERG_FQN)
print("iceberg BigFrames shape:", bdf.shape)   # (rows, cols) via metadata
# a bounded aggregate stays in BigQuery and returns a tiny result
print(bdf.groupby("payment_type")["trip_total"].mean().head(5).to_pandas())
```

### pandas + BigQuery Storage Read API (bounded)

`to_dataframe()` uses the Storage Read API (Arrow). We **bound** the pull (aggregate) rather than materialize 212M rows:

```python
sql = f'''
SELECT payment_type, COUNT(*) AS trips, AVG(trip_total) AS avg_total
FROM `{ICEBERG_FQN}`
GROUP BY payment_type
ORDER BY trips DESC
'''
df = bq.query(sql).to_dataframe()
print("rows returned:", len(df))
df.head()
```
