A quick tour of training and using machine learning models in BigQuery with SQL. BigQuery ML lets you `CREATE MODEL` and then evaluate, predict, and explain it with `ML.*` functions — all where your data already lives.

This notebook shows one example per category. For progressive, in-depth coverage, see the per-model folders under [`models/`](models/) and the [Detailed Reference](RESOURCES.md).

## Setup

```python
PROJECT_ID = 'statmike-mlops-349915'  # <-- Replace with your project ID
LOCATION = 'US'
DATASET_ID = 'bq_ml'
```

```python
try:
    from google.colab import auth
    auth.authenticate_user()
except ImportError:
    pass

from google.cloud import bigquery
client = bigquery.Client(project=PROJECT_ID)

dataset = bigquery.Dataset(bigquery.DatasetReference(PROJECT_ID, DATASET_ID))
dataset.location = LOCATION
client.create_dataset(dataset, exists_ok=True)
print(f'Dataset {PROJECT_ID}.{DATASET_ID} ready')

%load_ext bigquery_magics
```

---
## 1. Supervised Classification — `LOGISTIC_REG`

Train a binary classifier on census data (predict whether income exceeds $50K), then evaluate it — the core BigQuery ML loop. The full lifecycle (predict, explain, tune) is in [`models/logistic_regression/`](models/logistic_regression/).

```python
# Train (synchronous - completes when the model is ready)
query = f"""
CREATE OR REPLACE MODEL `{PROJECT_ID}.{DATASET_ID}.overview_logistic`
OPTIONS(model_type='LOGISTIC_REG', input_label_cols=['income_bracket'], auto_class_weights=TRUE) AS
SELECT age, education, occupation, hours_per_week, marital_status, income_bracket
FROM `bigquery-public-data.ml_datasets.census_adult_income`
"""
client.query(query).result()
print('Model trained')
```

```python
%%bigquery --project {PROJECT_ID}

SELECT *
FROM ML.EVALUATE(MODEL `statmike-mlops-349915.bq_ml.overview_logistic`)
```
