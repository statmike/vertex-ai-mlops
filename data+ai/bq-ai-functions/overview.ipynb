# BigQuery AI Functions — Overview

This notebook gives you a quick tour of BigQuery's AI functions — one working example per category so you can see what's available and decide where to dive deeper.

**Categories covered:**
1. **Generation** — Generate text and structured output with `AI.GENERATE`
2. **Managed Functions** — Classify, score, and filter with `AI.CLASSIFY`, `AI.SCORE`, `AI.IF`
3. **Structured Output** — Generate tabular data with `AI.GENERATE_TABLE`
4. **Embeddings & Search** — Create embeddings with `AI.EMBED` and search with `VECTOR_SEARCH`
5. **Forecasting** — Forecast time series with `AI.FORECAST`

**For deeper examples** of any function, see the individual function notebooks in the [`functions/`](functions/) folder.

**For setup details** (connections, models, quotas), see the [Setup Reference](setup/).

## 1. Generation — `AI.GENERATE`

`AI.GENERATE` is the simplest way to call a Gemini model from SQL. It's a **scalar function** — use it in SELECT, WHERE, or anywhere you'd use a regular function. No model creation required.

**Deep dive:** [`functions/ai_generate/`](functions/ai_generate/)

```python
%%bigquery --project {PROJECT_ID}

-- Simplest possible call: just a prompt string
SELECT
  city,
  (AI.GENERATE(CONCAT('What country is ', city, ' in? Answer in one word.'))).result AS country
FROM UNNEST(['Tokyo', 'Paris', 'Nairobi', 'Lima']) AS city
```

### Structured output with `output_schema`

`AI.GENERATE` can return typed columns instead of free-form text when you provide an `output_schema`.

```python
%%bigquery --project {PROJECT_ID}

-- Structured output: get typed columns back
SELECT
  animal,
  result.habitat,
  result.average_lifespan_years,
  result.is_endangered
FROM
  UNNEST(['Eagle', 'Salmon', 'Cobra']) AS animal,
  UNNEST([
    AI.GENERATE(
      CONCAT('Give me facts about: ', animal),
      output_schema => 'habitat STRING, average_lifespan_years INT64, is_endangered BOOL'
    )
  ]) AS result
```

## 2. Managed Functions — `AI.CLASSIFY`, `AI.SCORE`, `AI.IF`

These functions provide **simplified interfaces** for common tasks. BigQuery automatically optimizes the prompts sent to the model. They return simple scalar values (STRING, FLOAT64, BOOL) — not structs.

**Deep dives:** [`functions/ai_classify/`](functions/ai_classify/) | [`functions/ai_score/`](functions/ai_score/) | [`functions/ai_if/`](functions/ai_if/)

```python
%%bigquery --project {PROJECT_ID}

-- AI.CLASSIFY: assign a category from a list you provide
-- AI.SCORE: rate on a scale you describe
-- AI.IF: evaluate a natural language condition to TRUE/FALSE
SELECT
  review,
  AI.CLASSIFY(
    review,
    ['Product Quality', 'Shipping', 'Customer Service', 'Pricing']
  ) AS category,
  AI.SCORE(
    CONCAT('Rate the sentiment of this review on a scale from 1 (very negative) to 5 (very positive): ', review)
  ) AS sentiment_score,
  AI.IF(
    CONCAT(review, ' — Does this review mention a refund request?')
  ) AS mentions_refund
FROM UNNEST([
  'The product broke after two days. I want my money back!',
  'Shipping was incredibly fast, arrived next day.',
  'Great price for the quality. Will buy again.',
  'Customer support was unhelpful and rude.'
]) AS review
```

## 3. Structured Output — `AI.GENERATE_TABLE`

`AI.GENERATE_TABLE` is a **table-valued function** that returns structured columns defined by an `output_schema`. It requires a connection and remote model — both are created below.

This is particularly useful for **generating sample data** — which is how we use it throughout this project to keep examples self-contained.

**Deep dive:** [`functions/ai_generate_table/`](functions/ai_generate_table/)

```python
import subprocess, json

# Create connection (idempotent — needed for AI.GENERATE_TABLE)
subprocess.run(
    ['bq', 'mk', '--connection', '--location', LOCATION,
     '--connection_type', 'CLOUD_RESOURCE',
     '--project_id', PROJECT_ID, CONNECTION_ID],
    capture_output=True, text=True
)

# Grant Vertex AI User role to connection's service account (idempotent)
r = subprocess.run(
    ['bq', 'show', '--connection', '--format=json',
     '--project_id', PROJECT_ID, '--location', LOCATION, CONNECTION_ID],
    capture_output=True, text=True, check=True
)
sa = json.loads(r.stdout)['cloudResource']['serviceAccountId']
subprocess.run(
    ['gcloud', 'projects', 'add-iam-policy-binding', PROJECT_ID,
     f'--member=serviceAccount:{sa}', '--role=roles/aiplatform.user', '--quiet'],
    capture_output=True, text=True
)
print(f'Connection {CONNECTION_ID} ready (service account: {sa})')

# Create remote model (idempotent)
create_model_sql = f"""
CREATE OR REPLACE MODEL `{PROJECT_ID}.{DATASET_ID}.gemini_flash`
  REMOTE WITH CONNECTION `{PROJECT_ID}.{LOCATION}.{CONNECTION_ID}`
  OPTIONS (endpoint = 'gemini-2.5-flash')
"""
client.query(create_model_sql).result()
print('Model gemini_flash ready')
```

```python
query = f"""
SELECT ticket_id, customer_name, department, ticket_text
FROM AI.GENERATE_TABLE(
  MODEL `{PROJECT_ID}.{DATASET_ID}.gemini_flash`,
  (
    SELECT 'Generate 10 fictional customer support tickets for an online retailer. Include a mix of complaints, questions, and compliments. Make the text realistic and varied in length.' AS prompt
  ),
  STRUCT(
    'ticket_id INT64, customer_name STRING, ticket_text STRING, department STRING' AS output_schema
  )
)
"""
client.query(query).to_dataframe()
```

## 4. Embeddings & Search — `AI.EMBED` + `VECTOR_SEARCH`

`AI.EMBED` creates vector embeddings from text (scalar function, no model needed). `VECTOR_SEARCH` finds the nearest neighbors from pre-computed embeddings stored in a table.

**Deep dives:** [`functions/ai_embed/`](functions/ai_embed/) | [`functions/vector_search/`](functions/vector_search/)

```python
%%bigquery --project {PROJECT_ID}

-- AI.EMBED returns a vector embedding for each input text
SELECT
  doc,
  ARRAY_LENGTH(embedding) AS dimensions
FROM (
  SELECT
    doc,
    (AI.EMBED(
      content => doc,
      endpoint => 'text-embedding-005'
    )).result AS embedding
  FROM UNNEST([
    'BigQuery is a serverless data warehouse',
    'Cloud Storage stores unstructured data as objects',
    'Vertex AI provides ML model training and serving',
    'Cloud Run deploys containerized applications',
    'Pub/Sub is a messaging service for event-driven systems'
  ]) AS doc
)
```

### Semantic search with `VECTOR_SEARCH`

`VECTOR_SEARCH` finds the closest matches from pre-computed embeddings. It requires embeddings stored in a table — create one with `AI.EMBED`, then search it.

```python
# Step 1: Create a table with pre-computed embeddings
embed_sql = f"""
CREATE OR REPLACE TABLE `{PROJECT_ID}.{DATASET_ID}.overview_docs` AS
SELECT
  doc,
  (AI.EMBED(content => doc, endpoint => 'text-embedding-005')).result AS embedding
FROM UNNEST([
  'BigQuery is a serverless data warehouse',
  'Cloud Storage stores unstructured data as objects',
  'Vertex AI provides ML model training and serving',
  'Cloud Run deploys containerized applications',
  'Pub/Sub is a messaging service for event-driven systems'
]) AS doc
"""
client.query(embed_sql).result()
print('Embeddings table ready')

# Step 2: Search for the most relevant documents
search_sql = f"""
SELECT base.doc, distance
FROM VECTOR_SEARCH(
  TABLE `{PROJECT_ID}.{DATASET_ID}.overview_docs`,
  'embedding',
  (SELECT (AI.EMBED(
    content => 'How do I analyze large datasets?',
    endpoint => 'text-embedding-005'
  )).result AS embedding),
  top_k => 3,
  distance_type => 'COSINE'
)
ORDER BY distance
"""
client.query(search_sql).to_dataframe()
```

## 5. Forecasting — `AI.FORECAST`

`AI.FORECAST` uses Google's TimesFM foundation model to forecast time series data. No model training or creation needed — it works zero-shot on any time series.

**Deep dive:** [`functions/ai_forecast/`](functions/ai_forecast/)

```python
%%bigquery df_forecast --project {PROJECT_ID}

-- Generate a simple time series and forecast 7 days ahead
WITH daily_sales AS (
  SELECT
    DATE_ADD('2025-01-01', INTERVAL day_offset DAY) AS sale_date,
    -- Simulate sales with trend + weekly pattern + noise
    100 + (day_offset * 0.5)  -- upward trend
    + 20 * SIN(day_offset * 2 * ACOS(-1) / 7)  -- weekly cycle
    + (RAND() * 10 - 5)  -- noise
    AS daily_amount
  FROM UNNEST(GENERATE_ARRAY(0, 89)) AS day_offset
)
SELECT *
FROM AI.FORECAST(
  (SELECT * FROM daily_sales),
  data_col => 'daily_amount',
  timestamp_col => 'sale_date',
  horizon => 7,
  confidence_level => 0.95
)
```

## 6. Augmented Analytics — `AI.KEY_DRIVERS`

`AI.KEY_DRIVERS` performs key driver / contribution analysis: it finds the data segments that drive a change in a summable metric between an interest set and a reference set. No model, connection, or endpoint — just a table with a metric, dimensions, and a BOOL interest/reference column.

**Deep dive:** [`functions/ai_key_drivers/`](functions/ai_key_drivers/)

```python
%%bigquery --project {PROJECT_ID}

-- Why did total trip duration change from April 2016 to April 2017?
SELECT
  ARRAY_TO_STRING(drivers, ', ') AS segment,
  difference,
  contribution
FROM AI.KEY_DRIVERS(
  (SELECT
     tripduration,
     usertype,
     gender,
     (EXTRACT(YEAR FROM starttime) = 2017) AS is_interest
   FROM `bigquery-public-data.new_york_citibike.citibike_trips`
   WHERE EXTRACT(MONTH FROM starttime) = 4
     AND EXTRACT(YEAR FROM starttime) IN (2016, 2017)
     AND tripduration IS NOT NULL),
  metric_col => 'tripduration',
  dimension_cols => ['usertype', 'gender'],
  interest_label_col => 'is_interest',
  top_k => 8
)
ORDER BY contribution DESC
```
