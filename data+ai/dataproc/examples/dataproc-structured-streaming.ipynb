# Spark Structured Streaming for ML Inference on Dataproc

Spark [Structured Streaming](https://spark.apache.org/docs/latest/structured-streaming-programming-guide.html) with ML inference on a **provisioned Dataproc cluster**. Monitor a GCS directory for new JSON files, apply a sentiment model via Pandas UDF in `foreachBatch`, and write results to BigQuery.

**Why a provisioned cluster instead of Dataproc Serverless?**

Dataproc Serverless is designed for batch workloads — it auto-provisions executors, runs the job, and tears down. Structured Streaming needs a **long-running Spark session** that continuously polls for new data. Serverless jobs have a maximum runtime and will terminate the streaming query. A provisioned cluster stays alive as long as you need it, making it the right choice for streaming.

**Story arc:**
1. Understand Structured Streaming concepts (`readStream` / `writeStream`, triggers, checkpointing)
2. Create a provisioned Dataproc cluster (Serverless cannot do streaming)
3. Write a streaming PySpark script with ML inference via Pandas UDF in `foreachBatch`
4. Submit the job and simulate streaming data by uploading files to GCS
5. Verify results flow to BigQuery
6. Clean up the cluster and resources

## Understanding Structured Streaming

Spark [Structured Streaming](https://spark.apache.org/docs/latest/structured-streaming-programming-guide.html) treats a live data stream as an unbounded table that is continuously appended. Instead of `spark.read` (bounded), you use `spark.readStream` (unbounded). Instead of `df.write` (one-time), you use `df.writeStream` (continuous).

### Key Concepts

| Concept | Description |
|---------|-------------|
| `readStream` vs `read` | `readStream` creates an unbounded DataFrame that continuously monitors a source for new data. `read` loads a fixed dataset. |
| `writeStream` vs `write` | `writeStream` starts a continuous query that processes new data as it arrives. `write` saves once and stops. |
| **Trigger** | Controls when micro-batches execute. `processingTime` runs periodically, `once` runs a single batch, `availableNow` processes all available data then stops. |
| **Checkpoint** | A directory (typically on GCS) where Spark stores progress metadata. Enables exactly-once guarantees and recovery after restarts. |
| **Output mode** | `append` (new rows only), `complete` (full result table), `update` (changed rows only). |

### How It Works

1. The streaming query monitors an input source (GCS directory, Kafka topic, rate source)
2. At each trigger interval, Spark checks for new data since the last checkpoint
3. New data is processed as a micro-batch — a regular Spark DataFrame
4. Results are written to the output sink
5. Checkpoint is updated to record progress

### Comparison: Batch vs Structured Streaming vs Dataflow

| Aspect | Spark Batch (Notebook 2) | Spark Structured Streaming | Dataflow Streaming |
|--------|------------------------|---------------------------|-------------------|
| Processing model | Bounded dataset | Continuous micro-batch | True streaming |
| Data source | Static BQ tables | File streams, rate, Kafka | Pub/Sub |
| ML pattern | Pandas UDF | Pandas UDF in `foreachBatch` | RunInference |
| Cluster type | Serverless | Provisioned (required) | Serverless |
| Latency | Minutes | Seconds (trigger interval) | Milliseconds |
| Use case | Periodic batch scoring | Near-real-time enrichment | Event-driven predictions |

## Write the Streaming PySpark Script

This is the core of the notebook — a PySpark script that:

1. **`readStream`** from a GCS directory (JSON format) — Spark monitors the directory and picks up new files automatically
2. **`foreachBatch`** processes each micro-batch as a regular DataFrame — this is where ML inference happens
3. **Pandas UDF** with `_model_cache` applies sentiment analysis (same pattern as Notebook 2)
4. **BigQuery write** inside `foreachBatch` — the BigQuery Spark connector doesn't support direct `writeStream`, but works normally inside `foreachBatch` since each micro-batch is a regular DataFrame
5. **`processingTime` trigger** controls micro-batch frequency (every 30 seconds)
6. **Checkpointing** on GCS enables exactly-once guarantees
7. **`run_duration`** argument allows the job to self-terminate after a set time (useful for demos and testing)

### Why `foreachBatch`?

The BigQuery Spark connector supports `spark.read.format('bigquery')` and `df.write.format('bigquery')` but does **not** support `df.writeStream.format('bigquery')`. The `foreachBatch` sink bridges this gap — it receives each micro-batch as a regular DataFrame, so you can use `.write.format('bigquery')` inside it.

````python
%%writefile {DIR}/streaming_inference.py
"""Spark Structured Streaming with ML inference on Dataproc."""
import argparse
import os
import tempfile
import time

from pyspark.sql import SparkSession
from pyspark.sql.functions import pandas_udf, col, lit, current_timestamp
from pyspark.sql.types import StructType, StructField, StringType, FloatType, IntegerType
import pandas as pd

# Module-level model cache — persists across foreachBatch calls within the same executor
_model_cache = {}


def get_model(model_gcs_uri):
    """Load model from GCS, caching after first load."""
    if model_gcs_uri not in _model_cache:
        from transformers import pipeline as hf_pipeline, AutoTokenizer
        from google.cloud import storage

        client = storage.Client()
        parts = model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]
        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)
        # Load tokenizer explicitly — DistilBERT doesn't accept token_type_ids
        # but some transformers versions (e.g. runtime 2.2) generate them
        tokenizer = AutoTokenizer.from_pretrained(model_dir)
        if 'token_type_ids' in tokenizer.model_input_names:
            tokenizer.model_input_names.remove('token_type_ids')
        _model_cache[model_gcs_uri] = hf_pipeline(
            'sentiment-analysis', model=model_dir, tokenizer=tokenizer
        )
    return _model_cache[model_gcs_uri]


def run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--project', required=True)
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--model_gcs_uri', required=True)
    parser.add_argument('--model_name', default='distilbert')
    parser.add_argument('--temp_bucket', required=True)
    parser.add_argument('--input_path', required=True)
    parser.add_argument('--checkpoint_path', required=True)
    parser.add_argument('--trigger_interval', default='30 seconds')
    parser.add_argument('--run_duration', type=int, default=0,
                        help='Stop after N seconds (0=run forever)')
    args = parser.parse_args()

    spark = SparkSession.builder.appName('streaming-inference').getOrCreate()
    spark.conf.set('temporaryGcsBucket', args.temp_bucket)

    # Input schema for JSON files
    input_schema = StructType([
        StructField('id', IntegerType()),
        StructField('text', StringType()),
        StructField('category', StringType()),
    ])

    # Sentiment UDF — same pattern as batch inference (Notebook 2)
    model_uri = args.model_gcs_uri
    pred_schema = StructType([
        StructField('label', StringType()),
        StructField('score', FloatType()),
    ])

    @pandas_udf(pred_schema)
    def sentiment_udf(texts: pd.Series) -> pd.DataFrame:
        pipe = get_model(model_uri)
        results = pipe(texts.tolist(), batch_size=32)
        return pd.DataFrame(results)

    # Read stream from GCS — monitors the directory for new JSON files
    stream_df = spark.readStream.format('json') \
        .schema(input_schema) \
        .load(args.input_path)

    def process_batch(batch_df, batch_id):
        """Apply ML inference and write each micro-batch to BigQuery."""
        if batch_df.isEmpty():
            print(f'Batch {batch_id}: empty, skipping')
            return

        count = batch_df.count()
        print(f'Batch {batch_id}: processing {count} rows')

        # Apply sentiment model via Pandas UDF
        result_df = batch_df.withColumn('prediction', sentiment_udf(col('text')))
        output_df = result_df.select(
            col('id'),
            col('text'),
            col('category'),
            col('prediction.label').alias('label'),
            col('prediction.score').cast('float').alias('score'),
            lit(args.model_name).alias('model'),
            current_timestamp().alias('processed_at'),
        )

        # Write micro-batch to BigQuery using the standard connector
        output_df.write.format('bigquery') \
            .option('table', f'{args.project}.{args.bq_dataset}.{args.output_table}') \
            .mode('append') \
            .save()
        print(f'Batch {batch_id}: wrote {count} rows to BigQuery')

    # Start the streaming query
    query = stream_df.writeStream \
        .foreachBatch(process_batch) \
        .option('checkpointLocation', args.checkpoint_path) \
        .trigger(processingTime=args.trigger_interval) \
        .start()

    print(f'Streaming started. Trigger: {args.trigger_interval}')
    print(f'Input: {args.input_path}')
    print(f'Checkpoint: {args.checkpoint_path}')

    if args.run_duration > 0:
        print(f'Will stop after {args.run_duration} seconds')
        query.awaitTermination(timeout=args.run_duration * 1000)
        query.stop()
        print('Streaming stopped (duration reached)')
    else:
        query.awaitTermination()


if __name__ == '__main__':
    run()
````

## Prepare Streaming Input

Create JSON files in a GCS directory to simulate streaming data. The streaming job monitors this directory — each new file triggers a micro-batch.

We upload the first batch now. Additional batches will be uploaded after the streaming job starts, simulating data arriving over time.

````python
STREAMING_INPUT = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/streaming-input'
CHECKPOINT_PATH = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/checkpoints'
STREAMING_OUTPUT_TABLE = 'streaming_predictions'
````

````python
# Test sentences used throughout the notebook
SENTENCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
    "I cannot believe how amazing this product is, truly life-changing.",
    "Disappointing quality for the price. Would not recommend.",
    "The weather today is mild and partly cloudy.",
    "I had the worst customer service experience of my life yesterday.",
    "What a fantastic performance by the entire cast!",
    "The instructions were confusing and the product broke on day one.",
    "Pretty good overall, with some room for improvement.",
]

categories = ['entertainment', 'dining', 'general', 'product', 'product',
              'weather', 'service', 'entertainment', 'product', 'general']
````

````python
def upload_batch(batch_num, start_id, count):
    """Upload a batch of JSON records to GCS as a single file."""
    records = []
    for i in range(count):
        idx = (start_id + i - 1) % len(SENTENCES)
        records.append(json.dumps({
            'id': start_id + i,
            'text': SENTENCES[idx],
            'category': categories[idx],
        }))
    content = '\n'.join(records)
    blob_name = f'{SERIES}/{EXPERIMENT}/streaming-input/batch_{batch_num:03d}.json'
    bucket.blob(blob_name).upload_from_string(content, content_type='application/json')
    print(f'Uploaded batch {batch_num}: {count} records to gs://{GCS_BUCKET}/{blob_name}')
````

## Submit Streaming Job

Submit the streaming PySpark script to the provisioned cluster using `JobControllerClient`. This is different from Serverless:

| | Serverless (Notebook 2) | Provisioned (this notebook) |
|---|---|---|
| **Client** | `BatchControllerClient` | `JobControllerClient` |
| **Job type** | `Batch` / `PySparkBatch` | `Job` / `PySparkJob` |
| **Target** | Auto-provisioned runtime | Named cluster |
| **Submit method** | `create_batch()` | `submit_job()` |

The `--run_duration 300` argument tells the script to stop after 5 minutes. In production, you would omit this and let the streaming query run indefinitely.

````python
job = dataproc_v1.Job(
    placement=dataproc_v1.JobPlacement(cluster_name=CLUSTER_NAME),
    pyspark_job=dataproc_v1.PySparkJob(
        main_python_file_uri=streaming_script,
        args=[
            '--project', PROJECT_ID,
            '--bq_dataset', BQ_DATASET,
            '--output_table', STREAMING_OUTPUT_TABLE,
            '--model_gcs_uri', MODEL_URI,
            '--model_name', 'distilbert',
            '--temp_bucket', GCS_BUCKET,
            '--input_path', STREAMING_INPUT,
            '--checkpoint_path', CHECKPOINT_PATH,
            '--trigger_interval', '30 seconds',
            '--run_duration', '300',  # Run for 5 minutes then stop
        ],
        jar_file_uris=[
            'gs://spark-lib/bigquery/spark-bigquery-with-dependencies_2.12-0.36.1.jar',
        ],
    ),
)

print(f'Submitting streaming job to cluster {CLUSTER_NAME}...')
submitted_job = job_client.submit_job(
    request={
        'project_id': PROJECT_ID,
        'region': REGION,
        'job': job,
    }
)
job_id = submitted_job.reference.job_id
print(f'Job submitted: {job_id}')
print(f'Console: https://console.cloud.google.com/dataproc/jobs/{job_id}?region={REGION}&project={PROJECT_ID}')
````

## Trigger Modes

Structured Streaming supports several trigger modes that control when micro-batches execute:

| Trigger | Behavior | Use Case |
|---------|----------|----------|
| `processingTime='30 seconds'` | Process a micro-batch every 30 seconds | Production streaming |
| `once=True` | Process all available data in one micro-batch, then stop | Testing, backfill |
| `availableNow=True` | Process all available data (possibly multiple micro-batches), then stop | Incremental batch |

**`processingTime`** (used in this notebook) is the most common. The interval controls the trade-off between latency and throughput — shorter intervals mean lower latency but more overhead per micro-batch.

**`once`** is useful for testing: it processes everything available in a single micro-batch and stops. You can run a streaming job with `once=True` to test the pipeline without waiting for new data.

**`availableNow`** is a newer option that processes all available data but may split it across multiple micro-batches (respecting `maxFilesPerTrigger` and other limits). It is ideal for incremental batch processing — catch up on all new data since the last checkpoint, then stop.

```python
# Production: continuous processing
query = stream_df.writeStream \
    .foreachBatch(process_batch) \
    .trigger(processingTime='30 seconds') \
    .start()

# Testing: process once and stop
query = stream_df.writeStream \
    .foreachBatch(process_batch) \
    .trigger(once=True) \
    .start()

# Incremental batch: process all available, then stop
query = stream_df.writeStream \
    .foreachBatch(process_batch) \
    .trigger(availableNow=True) \
    .start()
```

## Summary

This notebook covered Spark Structured Streaming with ML inference on a provisioned Dataproc cluster:

1. **Provisioned cluster** — Structured Streaming needs a long-running Spark session. Dataproc Serverless is batch-only and terminates after the job completes.
2. **`foreachBatch`** — Bridges streaming and the BigQuery Spark connector. Each micro-batch is a regular DataFrame, so `.write.format('bigquery')` works inside it.
3. **Pandas UDF + `_model_cache`** — The same inference pattern from Notebook 2 works unchanged in streaming. The model loads once per executor and is reused across micro-batches.
4. **`processingTime` trigger** — Controls micro-batch frequency. 30-second intervals balance latency and throughput.
5. **Checkpointing** — GCS-based checkpoints enable exactly-once guarantees and recovery after restarts.

### When to Use Structured Streaming vs Dataflow

| Aspect | Spark Structured Streaming | Dataflow |
|--------|---------------------------|----------|
| Best for | Spark-native teams, file/Kafka sources | Event-driven, Pub/Sub sources |
| Cluster | You manage it (provisioned Dataproc) | Fully managed |
| Latency | Seconds (micro-batch) | Milliseconds (true streaming) |
| ML pattern | Pandas UDF in `foreachBatch` | RunInference API |
| Model hot-swap | Restart job or use custom logic | Built-in via side inputs |
