## Download Models to GCS

Download two HuggingFace sentiment models and upload them to GCS. Dataproc executors will download these model files from GCS at job startup.

- **v1 (DistilBERT):** Smaller, faster — primary model
- **v2 (BERT):** Larger — second model for multi-model comparison

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

## Pattern 1: Pandas UDF (Recommended)

A [Pandas UDF](https://spark.apache.org/docs/latest/api/python/user_guide/sql/arrow_pandas.html) is Spark's vectorized user-defined function. It receives and returns Pandas objects, with Arrow-optimized data transfer between JVM and Python.

**Why this is the recommended pattern for ML inference:**

- **Arrow-optimized:** Data transfers between Spark and Python use Apache Arrow columnar format — no row-by-row serialization overhead
- **Automatic batching:** Spark automatically groups rows into batches for the UDF, matching the batch-oriented nature of ML models
- **SQL integration:** Works with `withColumn`, `select`, and other DataFrame operations — the result is a struct column you can unpack
- **Structured output:** Returns a DataFrame with typed columns (`label: string`, `score: float`), not opaque tuples

**Module-level model caching:**

The key optimization is a module-level `_model_cache` dictionary. When Spark calls the UDF on a batch of rows, the model is loaded from GCS on the first call and cached. Subsequent UDF calls on the same executor reuse the cached model — no repeated downloads.

```python
_model_cache = {}

def get_model(model_gcs_uri):
    if model_gcs_uri not in _model_cache:
        # download from GCS, load model
        _model_cache[model_gcs_uri] = pipeline(...)
    return _model_cache[model_gcs_uri]
```

Without caching, every batch of rows would trigger a full model download from GCS — the model load time would dominate inference time.

````python
%%writefile {DIR}/inference_pandas_udf.py
"""Batch inference with Pandas UDF — the recommended pattern."""
import argparse
import os
import tempfile

from pyspark.sql import SparkSession
from pyspark.sql.functions import pandas_udf, col, lit, round as spark_round
from pyspark.sql.types import StructType, StructField, StringType, FloatType
import pandas as pd

# Module-level model cache — persists across UDF calls within the same executor
_model_cache = {}


def get_model(model_gcs_uri):
    """Load model from GCS, caching after first load per executor."""
    if model_gcs_uri not in _model_cache:
        from transformers import pipeline as hf_pipeline, AutoTokenizer
        from google.cloud import storage

        client = storage.Client()
        parts = model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]
        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)
        # Load tokenizer explicitly — DistilBERT doesn't accept token_type_ids
        # but some transformers versions (e.g. runtime 2.2) generate them
        tokenizer = AutoTokenizer.from_pretrained(model_dir)
        if 'token_type_ids' in tokenizer.model_input_names:
            tokenizer.model_input_names.remove('token_type_ids')
        _model_cache[model_gcs_uri] = hf_pipeline(
            'sentiment-analysis', model=model_dir, tokenizer=tokenizer
        )
    return _model_cache[model_gcs_uri]


def run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--project', required=True)
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--model_gcs_uri', required=True)
    parser.add_argument('--model_name', default='distilbert')
    parser.add_argument('--temp_bucket', required=True)
    args = parser.parse_args()

    spark = SparkSession.builder.appName('pandas-udf-inference').getOrCreate()
    spark.conf.set('temporaryGcsBucket', args.temp_bucket)

    # Read from BigQuery
    df = spark.read.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.source_table}') \
        .load() \
        .select('id', 'text', 'category')

    # Define sentiment UDF with struct return type
    model_uri = args.model_gcs_uri

    schema = StructType([
        StructField('label', StringType()),
        StructField('score', FloatType()),
    ])

    @pandas_udf(schema)
    def sentiment_udf(texts: pd.Series) -> pd.DataFrame:
        pipe = get_model(model_uri)
        results = pipe(texts.tolist(), batch_size=32)
        return pd.DataFrame(results)

    # Apply inference and format output
    result_df = df.withColumn('prediction', sentiment_udf(col('text')))
    output_df = result_df.select(
        col('id'),
        col('text'),
        col('category'),
        col('prediction.label').alias('label'),
        spark_round(col('prediction.score'), 6).alias('score'),
        lit(args.model_name).alias('model'),
    )

    # Write to BigQuery
    output_df.write.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.output_table}') \
        .mode('overwrite') \
        .save()

    print(f'Results written to {args.project}.{args.bq_dataset}.{args.output_table}')


if __name__ == '__main__':
    run()
````

## Pattern 2: mapPartitions (Alternative)

[`mapPartitions`](https://spark.apache.org/docs/latest/api/python/reference/api/pyspark.RDD.mapPartitions.html) is Spark's RDD-level transformation that gives you explicit control over how data is processed within each partition.

**How it differs from Pandas UDF:**

- **RDD-level:** Operates on the RDD API, not DataFrames — you iterate over rows manually
- **Per-partition model loading:** The model is loaded once per partition call (not via module-level caching, though you can combine both)
- **Row-by-row iteration:** You receive an iterator of Row objects and yield tuples
- **Returns tuples:** Output is an RDD of tuples that you convert back to a DataFrame with an explicit schema

**When to use mapPartitions instead of Pandas UDF:**

- You need explicit control over the model lifecycle (e.g., GPU allocation per partition)
- You want to implement custom batching logic within a partition
- Your output doesn't fit the struct-column pattern of Pandas UDFs
- You're working with an existing RDD-based codebase

````python
%%writefile {DIR}/inference_map_partitions.py
"""Batch inference with mapPartitions — lower-level alternative."""
import argparse
import os
import tempfile

from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import StructType, StructField, StringType, FloatType, IntegerType

# Module-level model cache — same pattern as Pandas UDF version
_model_cache = {}


def get_model(model_gcs_uri):
    """Load model from GCS, caching after first load per executor."""
    if model_gcs_uri not in _model_cache:
        from transformers import pipeline as hf_pipeline, AutoTokenizer
        from google.cloud import storage

        client = storage.Client()
        parts = model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]
        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)
        tokenizer = AutoTokenizer.from_pretrained(model_dir)
        if 'token_type_ids' in tokenizer.model_input_names:
            tokenizer.model_input_names.remove('token_type_ids')
        _model_cache[model_gcs_uri] = hf_pipeline(
            'sentiment-analysis', model=model_dir, tokenizer=tokenizer
        )
    return _model_cache[model_gcs_uri]


def run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--project', required=True)
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--model_gcs_uri', required=True)
    parser.add_argument('--model_name', default='distilbert')
    parser.add_argument('--temp_bucket', required=True)
    args = parser.parse_args()

    spark = SparkSession.builder.appName('map-partitions-inference').getOrCreate()
    spark.conf.set('temporaryGcsBucket', args.temp_bucket)

    # Read from BigQuery
    df = spark.read.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.source_table}') \
        .load() \
        .select('id', 'text', 'category')

    model_uri = args.model_gcs_uri
    model_name = args.model_name

    def predict_partition(iterator):
        """Process each partition: load model once, predict row by row."""
        pipe = get_model(model_uri)
        for row in iterator:
            result = pipe(row.text)[0]
            yield (
                row.id,
                row.text,
                row.category,
                result['label'],
                round(result['score'], 6),
                model_name,
            )

    output_schema = StructType([
        StructField('id', IntegerType()),
        StructField('text', StringType()),
        StructField('category', StringType()),
        StructField('label', StringType()),
        StructField('score', FloatType()),
        StructField('model', StringType()),
    ])

    # mapPartitions operates on the RDD, returns tuples
    results_rdd = df.rdd.mapPartitions(predict_partition)
    output_df = spark.createDataFrame(results_rdd, schema=output_schema)

    # Write to BigQuery
    output_df.write.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.output_table}') \
        .mode('overwrite') \
        .save()

    print(f'Results written to {args.project}.{args.bq_dataset}.{args.output_table}')


if __name__ == '__main__':
    run()
````

## Pattern Comparison

Both patterns produce the same results — the choice is about API level and control.

| Aspect | Pandas UDF | mapPartitions |
|--------|-----------|---------------|
| **API level** | DataFrame | RDD |
| **Batching** | Arrow-optimized, automatic | Row-by-row (manual batching possible) |
| **Return type** | Structured DataFrame (struct column) | Tuples → DataFrame |
| **SQL integration** | Yes (`withColumn`, `select`) | No (RDD-level) |
| **Model lifecycle** | Module-level cache (shared across UDF calls) | Per-partition explicit (one load per partition) |
| **Best for** | Most use cases — structured output, SQL-friendly | Full control over model lifecycle, custom batching |

**Recommendation:** Use Pandas UDF for most workloads. It's higher-level, Arrow-optimized, and integrates with DataFrame operations. Use mapPartitions only when you need explicit control over model loading or custom batching logic.

## Performance Tuning

Tuning Dataproc Serverless for ML inference involves balancing executor count against model memory. Each executor loads a separate copy of the model, so more executors means more parallelism but also more total memory.

### Key Configuration

| Config | Effect | Recommendation |
|--------|--------|----------------|
| `spark.executor.instances` | More executors = more parallel model copies | 2-4 for small models |
| `spark.executor.memory` | Must fit model + tokenizer + batch | 4g for DistilBERT, 8g for BERT |
| `batch_size` in UDF | Larger = more efficient inference | 32-64 for CPU |
| `spark.sql.execution.arrow.pyspark.enabled` | Arrow transfer for Pandas UDFs | true (default in 2.2) |

### Executor Count vs Model Memory

Each executor loads a full copy of the model into memory. For a 67M-parameter DistilBERT (~500MB in memory), 4 executors need ~2GB just for models. For BERT (~1GB), the same 4 executors need ~4GB.

```python
# Example: tuned batch for larger models
batch.runtime_config.properties.update({
    'spark.executor.instances': '4',
    'spark.executor.memory': '8g',
    'spark.driver.memory': '4g',
})
```

> **Driver memory constraint:** Dataproc Serverless enforces a minimum of 1024 MB/core for total memory (memory + 40% overhead). With the default 4 driver cores, `spark.driver.memory` must be at least 4g. Setting it lower (e.g., 2g) will cause the batch to fail at submission. The same constraint applies to executor memory and cores.

### Batch Size in the UDF

The `batch_size` parameter in the HuggingFace pipeline controls how many texts are tokenized and forwarded through the model at once. Larger batches use CPU/GPU more efficiently but require more memory.

```python
@pandas_udf(schema)
def sentiment_udf(texts: pd.Series) -> pd.DataFrame:
    pipe = get_model(model_uri)
    results = pipe(texts.tolist(), batch_size=64)  # increase for larger datasets
    return pd.DataFrame(results)
```

### Dynamic Allocation

Dataproc Serverless enables dynamic allocation by default — Spark adds or removes executors based on workload. For ML inference, this means:
- Small datasets may only use 1-2 executors (model loads on each)
- Large datasets will scale up, with each new executor loading its own model copy
- Set `spark.executor.instances` as a starting point; dynamic allocation adjusts from there

### Data Partitioning

The number of Spark partitions affects parallelism. With Pandas UDF, Spark processes one partition per UDF call. For small datasets (like our 10 rows), repartitioning is unnecessary. For large datasets:

```python
# Repartition to match executor count for even distribution
df = df.repartition(4)  # match spark.executor.instances
```

## Summary

This notebook covered two PySpark inference patterns on Dataproc Serverless:

1. **Pandas UDF (recommended)** — Arrow-optimized, DataFrame-level, returns structured output. Module-level `_model_cache` avoids repeated model downloads.
2. **mapPartitions (alternative)** — RDD-level, explicit model loading per partition, row-by-row iteration. Use when you need full control over the model lifecycle.
3. **Multi-model inference** — Submit separate batch jobs with different `--model_gcs_uri` args. Same script, different models.
4. **Performance tuning** — Balance executor count against model memory. Use `batch_size` in the UDF for efficient inference.

### Key Takeaways

- Pandas UDF is the recommended pattern for ML batch inference on Spark
- Module-level `_model_cache` avoids repeated model downloads per executor
- mapPartitions gives more control but is lower-level and row-oriented
- Multi-model inference via separate batch jobs, each with its own model URI
- Size executor memory to accommodate your model (~4g for DistilBERT, ~8g for BERT)
