# Calling Vertex AI Endpoints from Spark on Dataproc (Managed Service for Apache Spark)

Use Spark to call a deployed [Vertex AI Endpoint](https://cloud.google.com/vertex-ai/docs/predictions/overview) for ML inference. Instead of loading models onto Spark executors (as in the [batch inference](./dataproc-batch-inference.ipynb) and [streaming](./dataproc-structured-streaming.ipynb) notebooks), workers send prediction requests to a managed endpoint over the network.

**Two approaches to ML inference in Spark:**

| Approach | Where the model runs | Libraries needed on workers |
|----------|---------------------|----------------------------|
| **Model on Worker** ([batch](./dataproc-batch-inference.ipynb), [streaming](./dataproc-structured-streaming.ipynb)) | Spark executors | torch, transformers (Runtime 2.2) |
| **Model on Endpoint** (this notebook) | Vertex AI Endpoint | google-cloud-aiplatform only |

This is the same pattern used by AlloyDB and Spanner when they call Vertex AI Endpoints for SQL-based inference -- the data platform sends prediction requests to a centrally-managed model, rather than running the model locally. Spark executors become API callers, not model hosts.

**Story arc:**
1. Deploy a sentiment model to a Vertex AI Endpoint
2. Test the endpoint directly from the notebook
3. Write a PySpark script with a Pandas UDF that calls the endpoint
4. Submit a Dataproc Serverless batch job
5. Compare model-on-worker vs model-on-endpoint
6. Clean up

**Model:**

| Version | Model | Params | Labels |
|---------|-------|--------|--------|
| v1 | [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) | 67M | POSITIVE / NEGATIVE |

### Download Model to GCS

Download the DistilBERT model from HuggingFace and upload it to GCS. This uses the same download-and-upload pattern as the [batch inference notebook](./dataproc-batch-inference.ipynb).

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

MODEL_NAME = 'distilbert-base-uncased-finetuned-sst-2-english'
MODEL_GCS_PREFIX = f'{SERIES}/{EXPERIMENT}/models/v1'

existing = list(bucket.list_blobs(prefix=MODEL_GCS_PREFIX, max_results=1))
if existing:
    print(f'Model already in GCS at gs://{GCS_BUCKET}/{MODEL_GCS_PREFIX}')
else:
    print(f'Downloading {MODEL_NAME} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)
        tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{MODEL_GCS_PREFIX}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{MODEL_GCS_PREFIX}')

MODEL_ARTIFACT_URI = f'gs://{GCS_BUCKET}/{MODEL_GCS_PREFIX}'
print(f'Model artifact URI: {MODEL_ARTIFACT_URI}')
```

### Build Serving Container

Write a minimal FastAPI server that loads the model and serves predictions. The container:
1. Downloads model artifacts from GCS at startup (`AIP_STORAGE_URI` is a GCS URI, not a local path)
2. Loads the HuggingFace sentiment pipeline from the downloaded files
3. Exposes health and prediction routes via `AIP_HEALTH_ROUTE` and `AIP_PREDICT_ROUTE`

This is the same container pattern used across the [MLOps/Serving/Online](../../../MLOps/Serving/Online/) notebooks.

```python
SOURCE_DIR = f'{DIR}/source'
os.makedirs(SOURCE_DIR, exist_ok=True)
```

```python
%%writefile {DIR}/source/serve.py
"""Minimal FastAPI server for sentiment prediction on Vertex AI."""
import os
from fastapi import FastAPI, Request
from transformers import pipeline as hf_pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)
for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = hf_pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']
    texts = [inst['text'] for inst in instances]
    results = classifier(texts, batch_size=len(texts))
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
```

```python
%%writefile {DIR}/source/Dockerfile
FROM tiangolo/uvicorn-gunicorn-fastapi:python3.10

# Vertex AI expects the container on port 8080 (serving_container_ports)
ENV PORT 8080

COPY serve.py /app/main.py
COPY requirements.txt requirements.txt

RUN pip install --no-cache-dir --upgrade pip \
  && pip install --no-cache-dir -r requirements.txt
```

## Write the Endpoint-Calling PySpark Script

This PySpark script is fundamentally different from the model-on-worker inference scripts in the [batch inference](./dataproc-batch-inference.ipynb) and [streaming](./dataproc-structured-streaming.ipynb) notebooks:

| Aspect | Model on Worker (batch / streaming) | Model on Endpoint (this notebook) |
|--------|--------------------------------|-----------------------------------|
| **Imports** | torch, transformers | google-cloud-aiplatform |
| **Model loading** | Download from GCS, load into memory | None (model is on the endpoint) |
| **Inference** | `pipe(texts)` in-process | `endpoint.predict(instances)` over HTTP |
| **Runtime** | 2.2 required (ML libs) | Any runtime (only needs aiplatform SDK) |
| **Worker memory** | Must fit model (~500MB for DistilBERT) | Minimal (no model in memory) |

**Key design decision:** The Pandas UDF calls `aiplatform.Endpoint.predict()` which sends an HTTP request to the endpoint. Each UDF batch results in one API call. The model runs on the endpoint's dedicated compute, not on Spark executors.

**Batching:** The Pandas UDF naturally batches rows. Each batch becomes a single `predict()` call with multiple instances -- this is efficient because the endpoint processes them as a batch too.

```python
%%writefile {DIR}/endpoint_inference.py
"""PySpark inference via Vertex AI Endpoint -- no ML libs needed on workers."""
import argparse

from pyspark.sql import SparkSession
from pyspark.sql.functions import pandas_udf, col, lit, round as spark_round
from pyspark.sql.types import StructType, StructField, StringType, FloatType
import pandas as pd


def run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--project', required=True)
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--endpoint_id', required=True)
    parser.add_argument('--region', default='us-central1')
    parser.add_argument('--model_name', default='distilbert-endpoint')
    parser.add_argument('--temp_bucket', required=True)
    args = parser.parse_args()

    spark = SparkSession.builder.appName('endpoint-inference').getOrCreate()
    spark.conf.set('temporaryGcsBucket', args.temp_bucket)

    # Read from BigQuery
    df = spark.read.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.source_table}') \
        .load() \
        .select('id', 'text', 'category')

    # Capture args for use inside UDF (closures must reference simple values)
    endpoint_id = args.endpoint_id
    project = args.project
    region = args.region

    schema = StructType([
        StructField('label', StringType()),
        StructField('score', FloatType()),
    ])

    # Cache the endpoint client across UDF calls within the same executor
    _endpoint_cache = {}

    @pandas_udf(schema)
    def endpoint_udf(texts: pd.Series) -> pd.DataFrame:
        """Call Vertex AI Endpoint for inference."""
        if endpoint_id not in _endpoint_cache:
            from google.cloud import aiplatform
            aiplatform.init(project=project, location=region)
            _endpoint_cache[endpoint_id] = aiplatform.Endpoint(endpoint_id)

        ep = _endpoint_cache[endpoint_id]
        instances = [{'text': t} for t in texts.tolist()]
        response = ep.predict(instances=instances)

        results = [{'label': p['label'], 'score': p['score']} for p in response.predictions]
        return pd.DataFrame(results)

    # Apply inference and format output
    result_df = df.withColumn('prediction', endpoint_udf(col('text')))
    output_df = result_df.select(
        col('id'),
        col('text'),
        col('category'),
        col('prediction.label').alias('label'),
        spark_round(col('prediction.score'), 6).alias('score'),
        lit(args.model_name).alias('model'),
    )

    # Write to BigQuery
    output_df.write.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.output_table}') \
        .mode('overwrite') \
        .save()

    print(f'Results written to {args.project}.{args.bq_dataset}.{args.output_table}')


if __name__ == '__main__':
    run()
```

## Submit Dataproc Serverless Job

Upload the inference script to GCS and submit it as a Dataproc Serverless batch job.

**Key difference from the batch inference notebook:** The PySpark script does NOT import `torch` or `transformers`. It only needs `google-cloud-aiplatform`, which is pre-installed in Runtime 2.2. This means:

- No model download to each executor (faster startup)
- No model in executor memory (lower memory footprint)
- Workers only need network access to the Vertex AI Endpoint
- Runtime 2.2 is used for consistency, but any runtime with the aiplatform SDK would work

**Spark properties:** Dataproc Serverless enforces a minimum of 1024 MB per core for total memory (memory + 40% overhead). With the default 4 driver cores, `spark.driver.memory` must be at least `4g`. See [Notebook 1](./dataproc-serverless-fundamentals.ipynb) for details.

**Batch ID:** The batch ID is prefixed with `EXPERIMENT` so that listing and lifecycle operations can filter on it.

```python
bucket.blob(f'{SERIES}/{EXPERIMENT}/scripts/endpoint_inference.py').upload_from_filename(
    f'{DIR}/endpoint_inference.py'
)
script_path = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/scripts/endpoint_inference.py'
print(f'Script uploaded: {script_path}')
```

## Compare Approaches

The two patterns -- model on worker vs model on endpoint -- represent fundamentally different architectures for ML inference in data pipelines. Neither is universally better; the right choice depends on your workload.

### Architecture Comparison

| Aspect | Model on Worker ([batch](./dataproc-batch-inference.ipynb) / [streaming](./dataproc-structured-streaming.ipynb)) | Model on Endpoint (this notebook) |
|--------|--------------------------------|-----------------------------------|
| **Where model runs** | Spark executor process | Vertex AI Endpoint (separate infra) |
| **Runtime requirement** | 2.2 (needs torch, transformers) | Any (only needs aiplatform SDK) |
| **Inference latency** | In-process (microseconds) | Network round-trip (milliseconds) |
| **Model management** | GCS artifacts, manual versioning | Vertex AI Model Registry, traffic splitting |
| **Scaling** | Spark executors (coupled to data) | Endpoint replicas (independent of data) |
| **Worker memory** | Must fit model (~500MB-2GB per executor) | Minimal (no model in memory) |
| **Model updates** | Re-submit Dataproc job with new GCS path | Deploy new model version, no Spark changes |
| **Multi-consumer** | Each consumer loads its own copy | All consumers share one endpoint |
| **Cost** | Compute cost only | Compute + endpoint cost (always-on) |

### When to Use Each

**Use Model on Worker when:**
- High throughput batch jobs where in-process latency matters
- Cost-sensitive workloads (no always-on endpoint cost)
- Simple model management is acceptable (GCS paths)
- The model fits comfortably in executor memory
- Only Spark jobs consume the model

**Use Model on Endpoint when:**
- The model is already deployed for online serving (re-use the endpoint)
- The model is too large for executor memory (e.g., large language models)
- You need centralized model versioning and traffic splitting
- Multiple consumers (Spark, AlloyDB, Spanner, applications) share the model
- You want to update the model without changing Spark jobs
- You need A/B testing or canary deployments via endpoint traffic splitting

### The Broader Pattern

This "call an endpoint from a data platform" pattern appears across Google Cloud:

| Data Platform | How it calls the endpoint |
|---------------|--------------------------|
| **Spark on Dataproc** (this notebook) | Pandas UDF with `endpoint.predict()` |
| **AlloyDB** | `ML.PREDICT()` SQL function with remote model |
| **Spanner** | `ML.PREDICT()` SQL function with remote model |
| **BigQuery** | `ML.PREDICT()` with remote model or Vertex AI connection |

The model runs in one place (Vertex AI Endpoint), and every data platform calls it the same way -- over the network via a prediction API.

## Summary

This notebook demonstrated the **model-on-endpoint** pattern for ML inference in Spark -- calling a Vertex AI Endpoint from Dataproc Serverless instead of loading the model onto Spark executors.

### Key Takeaways

1. **No ML libraries on workers** -- The PySpark script only imports `google-cloud-aiplatform`, not `torch` or `transformers`. The model runs on the endpoint, not on executors.
2. **Same Pandas UDF pattern** -- The UDF structure is identical to the batch inference notebook. Only the inference call changes: `endpoint.predict()` instead of `pipe(texts)`.
3. **Centralized model management** -- The model lives in Vertex AI Model Registry. Update the model by deploying a new version to the endpoint -- no Spark job changes needed.
4. **Shared across platforms** -- The same endpoint can serve Spark, AlloyDB, Spanner, and application backends. One model, many consumers.
5. **Trade-off: latency vs management** -- In-process inference is faster, but endpoint inference gives you centralized versioning, traffic splitting, and multi-consumer access.
