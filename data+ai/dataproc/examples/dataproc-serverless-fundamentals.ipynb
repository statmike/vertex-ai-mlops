# Dataproc Serverless Fundamentals

Learn [Dataproc Serverless](https://cloud.google.com/dataproc-serverless/docs/overview) as a platform for running PySpark jobs on Google Cloud. No cluster to provision, no infrastructure to manage — submit a script, Dataproc auto-provisions executors, runs the job, and tears down.

**What makes Dataproc Serverless different from provisioned Dataproc clusters?**

| | Dataproc Serverless | Provisioned Dataproc Cluster |
|---|---|---|
| **Cluster management** | None — fully managed | You create, configure, and delete clusters |
| **Scaling** | Autoscales executors per job | Manual or autoscaling policies you configure |
| **Billing** | Pay only while the job runs | Pay while the cluster exists (even if idle) |
| **Startup time** | ~60s (shared pool) | Minutes (cluster creation) |
| **Use case** | Batch jobs, ETL, ML inference | Long-running workloads, streaming, GPUs |

**Story arc:** This notebook walks through the platform itself — how to write a PySpark script, submit it via the CLI and Python SDK, tune Spark properties, and monitor job lifecycle. ML inference is used as a motivating example at the end, but the focus is on Dataproc, not ML.

1. Create input data in BigQuery
2. Write a simple PySpark transform script
3. Submit via `gcloud` CLI and Python SDK
4. Tune Spark properties
5. Monitor job lifecycle
6. Preview: ML inference on Dataproc Serverless

## BigQuery Spark Connector

Dataproc Serverless includes the [BigQuery Spark connector](https://cloud.google.com/dataproc/docs/tutorials/bigquery-connector-spark-example) built in. This lets PySpark scripts read and write BigQuery tables directly:

**Read from BigQuery:**
```python
df = spark.read.format('bigquery') \
    .option('table', 'project.dataset.table') \
    .load()
```

**Write to BigQuery:**
```python
df.write.format('bigquery') \
    .option('table', 'project.dataset.output_table') \
    .mode('overwrite') \
    .save()
```

**Important:** The connector uses GCS as temporary storage for data transfer. You must set `temporaryGcsBucket` in your Spark session:
```python
spark.conf.set('temporaryGcsBucket', 'your-bucket')
```

## Write a PySpark Script — Simple Transform

Before diving into ML inference, let's start with a simple PySpark script that demonstrates the core Dataproc Serverless workflow:

1. Read from BigQuery using the Spark BigQuery connector
2. Apply a transformation (add uppercase category and word count columns)
3. Write results back to a new BigQuery table

This teaches the BigQuery I/O pattern without the complexity of ML model loading.

````python
%%writefile {DIR}/transform.py
"""Simple PySpark transform — reads from BigQuery, adds columns, writes back."""
import argparse
from pyspark.sql import SparkSession
from pyspark.sql.functions import upper, col, size, split


def run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--project', required=True)
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--temp_bucket', required=True)
    args = parser.parse_args()

    spark = SparkSession.builder.appName('simple-transform').getOrCreate()
    spark.conf.set('temporaryGcsBucket', args.temp_bucket)

    df = spark.read.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.source_table}') \
        .load()

    result = df.select(
        col('id'),
        col('text'),
        upper(col('category')).alias('category_upper'),
        size(split(col('text'), ' ')).alias('word_count'),
    )

    result.write.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.output_table}') \
        .mode('overwrite') \
        .save()

    print(f'Results written to {args.project}.{args.bq_dataset}.{args.output_table}')


if __name__ == '__main__':
    run()
````

````python
bucket.blob(f'{SERIES}/{EXPERIMENT}/scripts/transform.py').upload_from_filename(f'{DIR}/transform.py')
transform_script = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/scripts/transform.py'
print(f'Script uploaded: {transform_script}')
````

## Submit via Python SDK

For programmatic control, use the `google-cloud-dataproc` Python SDK. The key components:

- **`BatchControllerClient`** — manages batch job submission and monitoring
- **`Batch`** — the job definition (script, args, runtime)
- **`PySparkBatch`** — PySpark-specific settings (main script, jar files, Python files)
- **`RuntimeConfig`** — runtime version and Spark properties

The SDK returns a long-running operation that you can poll or block on with `.result()`.

````python
batch_id = f'transform-sdk-{datetime.now().strftime("%Y%m%d-%H%M%S")}'

batch = dataproc_v1.Batch(
    pyspark_batch=dataproc_v1.PySparkBatch(
        main_python_file_uri=transform_script,
        args=[
            '--project', PROJECT_ID,
            '--bq_dataset', BQ_DATASET,
            '--source_table', 'input_data',
            '--output_table', 'transform_sdk_output',
            '--temp_bucket', GCS_BUCKET,
        ],
    ),
    runtime_config=dataproc_v1.RuntimeConfig(version='2.2'),
)

print(f'Submitting batch job: {batch_id}')
operation = dataproc_client.create_batch(
    request={
        'parent': f'projects/{PROJECT_ID}/locations/{REGION}',
        'batch': batch,
        'batch_id': batch_id,
    }
)
print(f'Console: https://console.cloud.google.com/dataproc/batches/{REGION}/{batch_id}?project={PROJECT_ID}')
````

````python
response = operation.result()
print(f'Batch state: {dataproc_v1.Batch.State(response.state).name}')
````

````python
bq_client.query(f'SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.transform_sdk_output` ORDER BY id').to_dataframe()
````

## Spark Properties

Dataproc Serverless auto-provisions resources, but you can tune the Spark configuration via properties. These control executor count, memory, and autoscaling behavior.

| Property | Description | Default |
|----------|-------------|----------|
| `spark.executor.instances` | Number of executors | 2 |
| `spark.executor.memory` | Memory per executor | 1g |
| `spark.executor.cores` | Cores per executor | 2 |
| `spark.driver.memory` | Driver memory | 1g |
| `spark.dynamicAllocation.enabled` | Enable autoscaling | true |

Properties are set via `runtime_config.properties` in the SDK or `--properties` in the CLI. The example below submits the same transform with more memory and executors.

````python
batch_id_tuned = f'transform-tuned-{datetime.now().strftime("%Y%m%d-%H%M%S")}'

batch_tuned = dataproc_v1.Batch(
    pyspark_batch=dataproc_v1.PySparkBatch(
        main_python_file_uri=transform_script,
        args=[
            '--project', PROJECT_ID,
            '--bq_dataset', BQ_DATASET,
            '--source_table', 'input_data',
            '--output_table', 'transform_tuned_output',
            '--temp_bucket', GCS_BUCKET,
        ],
    ),
    runtime_config=dataproc_v1.RuntimeConfig(version='2.2'),
    environment_config=dataproc_v1.EnvironmentConfig(
        execution_config=dataproc_v1.ExecutionConfig(
            subnetwork_uri=f'projects/{PROJECT_ID}/regions/{REGION}/subnetworks/default',
        ),
    ),
)
batch_tuned.runtime_config.properties.update({
    'spark.executor.instances': '4',
    'spark.executor.memory': '4g',
    'spark.driver.memory': '2g',
})

print(f'Submitting tuned batch job: {batch_id_tuned}')
operation_tuned = dataproc_client.create_batch(
    request={
        'parent': f'projects/{PROJECT_ID}/locations/{REGION}',
        'batch': batch_tuned,
        'batch_id': batch_id_tuned,
    }
)
response_tuned = operation_tuned.result()
print(f'Batch state: {dataproc_v1.Batch.State(response_tuned.state).name}')
````

## Job Lifecycle

Each Dataproc Serverless batch job goes through a defined set of states:

| State | Description |
|-------|-------------|
| `PENDING` | Job accepted, waiting for resources |
| `RUNNING` | Executors provisioned, Spark job executing |
| `SUCCEEDED` | Job completed successfully |
| `FAILED` | Job failed (check logs for details) |
| `CANCELLED` | Job was cancelled before completion |

The SDK provides methods to list, get details, and cancel batch jobs.

````python
# List recent batches
print('Recent batches:')
print(f'{"Batch ID":<50} {"State":<12} {"Created"}')
print('-' * 85)
for batch_info in dataproc_client.list_batches(
    request={'parent': f'projects/{PROJECT_ID}/locations/{REGION}'}
):
    state = dataproc_v1.Batch.State(batch_info.state).name
    created = batch_info.create_time.strftime('%Y-%m-%d %H:%M:%S') if batch_info.create_time else 'N/A'
    name = batch_info.name.split('/')[-1]
    if EXPERIMENT in name:
        print(f'{name:<50} {state:<12} {created}')
````

````python
# Get batch details
batch_detail = dataproc_client.get_batch(
    request={'name': f'projects/{PROJECT_ID}/locations/{REGION}/batches/{batch_id}'}
)
print(f'Batch: {batch_detail.name.split("/")[-1]}')
print(f'State: {dataproc_v1.Batch.State(batch_detail.state).name}')
print(f'Created: {batch_detail.create_time}')
print(f'Runtime: {batch_detail.runtime_config.version}')
````

## Custom Containers

For most ML workloads, runtime 2.2 has everything you need. But when you need full control over the environment — pinned library versions, additional native libraries, or private packages — use a custom container.

**When to use custom containers:**
- Pinned versions of torch, transformers, etc. for reproducibility
- Native libraries (e.g., OpenCV, FFmpeg) not in the default runtime
- Private PyPI packages from your internal registry
- Specific CUDA versions for GPU workloads

**What they must NOT include:**
- Apache Spark — it's mounted at runtime by Dataproc. Including Spark in your container will cause conflicts.

**How to specify:**
```python
batch = dataproc_v1.Batch(
    pyspark_batch=dataproc_v1.PySparkBatch(
        main_python_file_uri='gs://bucket/script.py',
    ),
    runtime_config=dataproc_v1.RuntimeConfig(
        container_image='us-central1-docker.pkg.dev/project/repo/image:tag',
    ),
)
```

See the [custom container documentation](https://cloud.google.com/dataproc-serverless/docs/guides/custom-containers) for Dockerfile examples and requirements.

### Download Models to GCS

Download two HuggingFace sentiment models and upload them to GCS. Dataproc executors will download these model files from GCS at job startup.

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

### Write the Inference Script

This PySpark script uses a Pandas UDF to run HuggingFace sentiment analysis. The model is loaded from GCS once per executor and cached in a module-level dictionary for reuse across UDF calls.

````python
%%writefile {DIR}/inference.py
"""PySpark sentiment inference with Pandas UDF on Dataproc Serverless."""
import argparse
import os
import tempfile

from pyspark.sql import SparkSession
from pyspark.sql.functions import pandas_udf, col, lit, round as spark_round
from pyspark.sql.types import StructType, StructField, StringType, FloatType
import pandas as pd

# Module-level model cache — persists across UDF calls within the same executor
_model_cache = {}


def get_model(model_gcs_uri):
    """Load model from GCS, caching after first load per executor."""
    if model_gcs_uri not in _model_cache:
        from transformers import pipeline as hf_pipeline, AutoTokenizer
        from google.cloud import storage

        client = storage.Client()
        parts = model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]
        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)
        # Load tokenizer explicitly — DistilBERT doesn't accept token_type_ids
        # but some transformers versions (e.g. runtime 2.2) generate them
        tokenizer = AutoTokenizer.from_pretrained(model_dir)
        if 'token_type_ids' in tokenizer.model_input_names:
            tokenizer.model_input_names.remove('token_type_ids')
        _model_cache[model_gcs_uri] = hf_pipeline(
            'sentiment-analysis', model=model_dir, tokenizer=tokenizer
        )
    return _model_cache[model_gcs_uri]


def run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--project', required=True)
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--model_gcs_uri', required=True)
    parser.add_argument('--model_name', default='distilbert')
    parser.add_argument('--temp_bucket', required=True)
    args = parser.parse_args()

    spark = SparkSession.builder.appName('batch-inference').getOrCreate()
    spark.conf.set('temporaryGcsBucket', args.temp_bucket)

    # Read from BigQuery
    df = spark.read.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.source_table}') \
        .load() \
        .select('id', 'text', 'category')

    # Define sentiment UDF with struct return type
    model_uri = args.model_gcs_uri

    schema = StructType([
        StructField('label', StringType()),
        StructField('score', FloatType()),
    ])

    @pandas_udf(schema)
    def sentiment_udf(texts: pd.Series) -> pd.DataFrame:
        pipe = get_model(model_uri)
        results = pipe(texts.tolist(), batch_size=32)
        return pd.DataFrame(results)

    # Apply inference and format output
    result_df = df.withColumn('prediction', sentiment_udf(col('text')))
    output_df = result_df.select(
        col('id'),
        col('text'),
        col('prediction.label').alias('label'),
        spark_round(col('prediction.score'), 6).alias('score'),
        lit(args.model_name).alias('model'),
    )

    # Write to BigQuery
    output_df.write.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.output_table}') \
        .mode('overwrite') \
        .save()

    print(f'Results written to {args.project}.{args.bq_dataset}.{args.output_table}')


if __name__ == '__main__':
    run()
````

### Submit ML Inference Job

Upload the inference script to GCS and submit it as a Dataproc Serverless batch job — the same pattern as the simple transform, but now running HuggingFace model inference.

````python
bucket.blob(f'{SERIES}/{EXPERIMENT}/scripts/inference.py').upload_from_filename(f'{DIR}/inference.py')
inference_script = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/scripts/inference.py'
print(f'Script uploaded: {inference_script}')
````

````python
BQ_INFERENCE_TABLE = 'inference_predictions'
batch_id_ml = f'inference-{datetime.now().strftime("%Y%m%d-%H%M%S")}'

batch_ml = dataproc_v1.Batch(
    pyspark_batch=dataproc_v1.PySparkBatch(
        main_python_file_uri=inference_script,
        args=[
            '--project', PROJECT_ID,
            '--bq_dataset', BQ_DATASET,
            '--source_table', 'input_data',
            '--output_table', BQ_INFERENCE_TABLE,
            '--model_gcs_uri', MODELS['v1']['artifact_uri'],
            '--model_name', 'distilbert',
            '--temp_bucket', GCS_BUCKET,
        ],
    ),
    runtime_config=dataproc_v1.RuntimeConfig(version='2.2'),
)

print(f'Submitting ML inference job: {batch_id_ml}')
operation_ml = dataproc_client.create_batch(
    request={
        'parent': f'projects/{PROJECT_ID}/locations/{REGION}',
        'batch': batch_ml,
        'batch_id': batch_id_ml,
    }
)
print(f'Console: https://console.cloud.google.com/dataproc/batches/{REGION}/{batch_id_ml}?project={PROJECT_ID}')
````

## Summary

This notebook covered the Dataproc Serverless platform fundamentals:

1. **Runtimes** — Runtime 2.2 includes torch, transformers, and other ML libraries pre-installed
2. **BigQuery Spark connector** — `spark.read.format('bigquery')` and `.write.format('bigquery')` with `temporaryGcsBucket`
3. **Job submission** — `gcloud` CLI for quick iteration, Python SDK (`BatchControllerClient`) for programmatic control
4. **Spark properties** — Tune executor count, memory, and autoscaling via `runtime_config.properties`
5. **Job lifecycle** — List, inspect, and monitor batch jobs through PENDING → RUNNING → SUCCEEDED states
6. **Custom containers** — For pinned versions or native libraries, but must NOT include Spark
7. **ML inference** — Pandas UDF with module-level model caching, same submit pattern as any PySpark job

### Decision Guide

| Need | Use |
|------|-----|
| No cluster management, ML libs pre-installed | Dataproc Serverless Runtime 2.2 |
| Custom dependencies or pinned versions | Dataproc Serverless + custom container |
| Long-running streaming jobs | Provisioned Dataproc cluster |
| Full cluster control, GPUs | Provisioned Dataproc cluster |
