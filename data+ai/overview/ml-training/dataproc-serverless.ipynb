# Dataproc Serverless — Distributed Processing Without Cluster Management

This notebook demonstrates [Dataproc Serverless](https://cloud.google.com/dataproc-serverless/docs/overview) — running PySpark and SparkR workloads at any scale with zero cluster provisioning. Infrastructure spins up for your job and shuts down when it finishes.

Dataproc Serverless offers two modes:
- **Interactive sessions** — Jupyter-connected Spark sessions for iterative development and debugging
- **Batch jobs** — submit standalone scripts that auto-scale and run to completion

This notebook covers both, plus custom containers for adding proprietary libraries and additional packages.

**What it demonstrates:**
- Creating a Dataproc Serverless interactive session via the API
- PySpark MLlib on BigQuery data (logistic regression for fraud detection)
- Packaging PySpark logic as a standalone script and submitting it as a batch job
- Building custom container images for Dataproc Serverless
- SparkR workloads on the same serverless infrastructure

**The story:** The sibling `bigframes-local.ipynb` trains locally, `bqml.ipynb` trains in-database with SQL, and this notebook trains with distributed Spark — same fraud detection task, different compute paradigm. Dataproc Serverless is the path when your data or computation exceeds what a single machine can handle, or when you need the full Spark/SparkR ecosystem.

**Dataset:** [`bigquery-public-data.ml_datasets.ulb_fraud_detection`](https://console.cloud.google.com/bigquery?p=bigquery-public-data&d=ml_datasets&t=ulb_fraud_detection) — 284,807 credit card transactions with 492 frauds (0.17%). Features `V1`–`V28` are PCA-transformed, plus `Time`, `Amount`, and binary `Class` (0=legitimate, 1=fraud).

### Connect to a Serverless Spark Session

`DataprocSparkSession` handles session creation and Spark Connect setup in one step. The `.config()` calls set Spark properties that control the cluster shape — executor count, cores, and memory.

````python
spark = (DataprocSparkSession.builder
    .appName('fraud-exploration')
    .projectId(PROJECT_ID)
    .location(REGION)
    .config('spark.executor.instances', '4')
    .config('spark.executor.cores', '4')
    .config('spark.executor.memory', '4g')
    .config('spark.driver.memory', '4g')
    .getOrCreate())

print('Spark session ready — connected to Dataproc Serverless via Spark Connect')
````

````python
df = (spark.read.format('bigquery')
    .option('table', BQ_SOURCE)
    .load())

row_count = df.count()
print(f'Loaded {row_count:,} rows, {len(df.columns)} columns')

# Repartition to spread data across all executor cores.
# BigQuery Storage API may produce few partitions — with 16 cores most would sit idle.
# Rule of thumb: 2-4 partitions per core → 32-64 for 16 cores
df = df.repartition(48)
df.cache()
df.count()
print(f'Repartitioned to 48 partitions — data spread across all 4 executors × 4 cores')
````

### Executor Configuration and Data Rebalancing

**Executor configuration** controls the cluster shape — how many machines, how many cores each, and how much memory. This determines how much parallelism Spark can use.

| Property | What It Controls | Example |
|---|---|---|
| `spark.executor.instances` | Number of executor JVMs | `4` — four worker processes |
| `spark.executor.cores` | CPU cores per executor | `4` — each executor uses 4 cores |
| `spark.executor.memory` | RAM per executor | `4g` — 4 GB heap per executor |
| `spark.driver.memory` | RAM for the driver | `4g` — driver coordinates the job |
| `spark.dynamicAllocation.minExecutors` | Minimum executors when idle | `2` |
| `spark.dynamicAllocation.maxExecutors` | Maximum executors under load | `20` |

With 4 executors × 4 cores, Spark can run **16 tasks in parallel**. But parallelism only helps if the data is spread across enough partitions.

**Data rebalancing** ensures data is evenly distributed across executors:

| Operation | What It Does | When to Use |
|---|---|---|
| `df.repartition(n)` | Shuffle data into `n` partitions (even distribution) | After reading from BigQuery (which may produce few partitions), before compute-heavy operations |
| `df.coalesce(n)` | Reduce to `n` partitions without a full shuffle | After filtering that shrinks the DataFrame significantly |

**Why this matters:** When reading from BigQuery via the Storage API, Spark may receive data in just a few partitions. If you have 16 cores but only 3 partitions, 13 cores sit idle. A good rule of thumb: **2–4 partitions per executor core** (so 32–64 partitions for our 16-core setup).

> **Note:** `df.rdd.getNumPartitions()` can inspect partition count in classic PySpark, but the RDD API is not available via Spark Connect. In batch jobs (which use a full SparkSession), this works — see the batch script below.

````python
from pyspark.sql.functions import when, col
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml import Pipeline

feature_cols = [f'V{i}' for i in range(1, 29)] + ['Amount']
assembler = VectorAssembler(inputCols=feature_cols, outputCol='raw_features')
scaler = StandardScaler(inputCol='raw_features', outputCol='features', withStd=True, withMean=False)

# Class weights for the 578:1 imbalance
neg = df.filter('Class = 0').count()
pos = df.filter('Class = 1').count()
weight_ratio = neg / pos
weighted = df.withColumn('weight', when(col('Class') == 1, weight_ratio).otherwise(1.0))

train, test = weighted.randomSplit([0.8, 0.2], seed=42)
print(f'Train: {train.count():,}, Test: {test.count():,}, Weight ratio: {weight_ratio:.0f}:1')

evaluator = BinaryClassificationEvaluator(labelCol='Class', metricName='areaUnderROC')
````

````python
lr = LogisticRegression(featuresCol='features', labelCol='Class', weightCol='weight', maxIter=100)
lr_pipeline = Pipeline(stages=[assembler, scaler, lr])

print('Training Logistic Regression...')
lr_model = lr_pipeline.fit(train)
lr_auc = evaluator.evaluate(lr_model.transform(test))
print(f'Logistic Regression AUC-ROC: {lr_auc:.4f}')
````

````python
rf = RandomForestClassifier(
    featuresCol='features', labelCol='Class', weightCol='weight',
    numTrees=20, maxDepth=6, seed=42,
)
rf_pipeline = Pipeline(stages=[assembler, scaler, rf])

print('Training Random Forest...')
rf_model = rf_pipeline.fit(train)
rf_auc = evaluator.evaluate(rf_model.transform(test))
print(f'Random Forest AUC-ROC: {rf_auc:.4f}')
````

````python
spark.stop()
print('Spark session stopped — serverless infrastructure shuts down automatically.')
````

## Batch Execution — Production PySpark Jobs

Once your PySpark logic is proven interactively, you package it as a standalone `.py` script and submit it as a [Dataproc Serverless batch job](https://cloud.google.com/dataproc-serverless/docs/concepts/batch). The batch infrastructure auto-scales, runs the script to completion, and shuts down — no cluster to manage.

**The pattern:**
1. Define the PySpark script as a string
2. Upload it to GCS
3. Submit it via `BatchControllerClient`
4. Poll for completion
5. Read results from BigQuery

This is the production workflow for Spark on Google Cloud — same code you prototyped interactively, now running at scale with automatic resource management.

### The PySpark Script

This script reads the fraud detection dataset from BigQuery, trains three models with PySpark MLlib, evaluates them, and writes predictions and metrics back to BigQuery.

**Package installation:** The script installs `xgboost` at startup using `pip install` — the same pattern as the `vertex-training.ipynb` training scripts. XGBoost provides a [`SparkXGBClassifier`](https://xgboost.readthedocs.io/en/latest/python/python_api.html#module-xgboost.spark) that integrates directly with PySpark ML Pipelines. For production workloads, bake dependencies into a custom container instead (see the Custom Containers section below).

````python
SPARK_SCRIPT = '''
#!/usr/bin/env python3
"""PySpark fraud detection — three models on Dataproc Serverless Batch."""
import subprocess, sys

# Install additional packages — not in the default Dataproc runtime.
# For production, bake these into a custom container instead.
subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', 'xgboost'])

from pyspark.sql import SparkSession
from pyspark.sql.functions import when, col
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.ml import Pipeline
from xgboost.spark import SparkXGBClassifier

def main():
    project_id = sys.argv[1]
    bq_dataset = sys.argv[2]
    bq_source = sys.argv[3]

    spark = SparkSession.builder.appName("fraud-detection-batch").getOrCreate()

    # --- Read from BigQuery ---
    df = (spark.read.format("bigquery")
          .option("table", bq_source)
          .load())
    n = df.count()
    fraud_count = df.filter("Class = 1").count()
    print(f"Loaded {n:,} rows — {fraud_count} frauds ({fraud_count/n*100:.2f}%)")
    print(f"Initial partitions: {df.rdd.getNumPartitions()}")

    # --- Repartition for parallelism ---
    # BigQuery Storage API may produce few partitions.
    # Repartition to spread data across all executor cores.
    df = df.repartition(48)
    df.cache()
    df.count()  # Materialize the cache
    print(f"Repartitioned to {df.rdd.getNumPartitions()} partitions")

    # --- Feature engineering ---
    feature_cols = [f"V{i}" for i in range(1, 29)] + ["Amount"]
    assembler = VectorAssembler(inputCols=feature_cols, outputCol="raw_features")
    scaler = StandardScaler(inputCol="raw_features", outputCol="features",
                            withStd=True, withMean=False)

    # --- Class weights for imbalance ---
    neg_count = df.filter("Class = 0").count()
    pos_count = df.filter("Class = 1").count()
    weight_ratio = neg_count / pos_count
    df = df.withColumn("weight", when(col("Class") == 1, weight_ratio).otherwise(1.0))
    print(f"Class weight ratio: {weight_ratio:.1f}:1")

    # --- Train/test split ---
    train, test = df.randomSplit([0.8, 0.2], seed=42)
    print(f"Train: {train.count():,} rows, Test: {test.count():,} rows")

    # --- Logistic Regression pipeline ---
    lr = LogisticRegression(
        featuresCol="features", labelCol="Class", weightCol="weight",
        maxIter=100, regParam=0.01, elasticNetParam=0.5
    )
    lr_pipeline = Pipeline(stages=[assembler, scaler, lr])
    print("Training Logistic Regression...")
    lr_model = lr_pipeline.fit(train)

    # --- Random Forest pipeline ---
    rf = RandomForestClassifier(
        featuresCol="features", labelCol="Class", weightCol="weight",
        numTrees=100, maxDepth=10, seed=42
    )
    rf_pipeline = Pipeline(stages=[assembler, scaler, rf])
    print("Training Random Forest...")
    rf_model = rf_pipeline.fit(train)

    # --- XGBoost pipeline (installed via pip) ---
    xgb = SparkXGBClassifier(
        features_col="features", label_col="Class", weight_col="weight",
        n_estimators=100, max_depth=6, learning_rate=0.1,
        scale_pos_weight=weight_ratio, eval_metric="aucpr",
        verbosity=0,
    )
    xgb_pipeline = Pipeline(stages=[assembler, scaler, xgb])
    print("Training XGBoost (pip-installed)...")
    xgb_model = xgb_pipeline.fit(train)

    # --- Evaluate all three models ---
    auc_eval = BinaryClassificationEvaluator(labelCol="Class", metricName="areaUnderROC")
    pr_eval = BinaryClassificationEvaluator(labelCol="Class", metricName="areaUnderPR")
    f1_eval = MulticlassClassificationEvaluator(labelCol="Class", metricName="f1")

    results = []
    for name, model in [("LogisticRegression", lr_model),
                        ("RandomForest", rf_model),
                        ("XGBoost", xgb_model)]:
        preds = model.transform(test)
        auc = auc_eval.evaluate(preds)
        pr = pr_eval.evaluate(preds)
        f1 = f1_eval.evaluate(preds)
        print(f"{name}: AUC-ROC={auc:.4f}, AUC-PR={pr:.4f}, F1={f1:.4f}")
        results.append((name, round(auc, 6), round(pr, 6), round(f1, 6)))

    # --- Write metrics to BigQuery ---
    metrics_df = spark.createDataFrame(results, ["model", "auc_roc", "auc_pr", "f1"])
    metrics_table = f"{project_id}.{bq_dataset}.spark_fraud_metrics"
    (metrics_df.write.format("bigquery")
     .option("table", metrics_table)
     .option("writeMethod", "direct")
     .mode("overwrite")
     .save())
    print(f"Metrics written to {metrics_table}")

    # --- Write best model predictions to BigQuery ---
    best_name = max(results, key=lambda x: x[1])[0]
    models = {"LogisticRegression": lr_model, "RandomForest": rf_model, "XGBoost": xgb_model}
    best_model = models[best_name]
    print(f"Best model by AUC-ROC: {best_name}")

    best_preds = best_model.transform(test).select(
        "Time", "Amount", "Class", "prediction", "probability"
    )
    preds_table = f"{project_id}.{bq_dataset}.spark_fraud_predictions"
    (best_preds.write.format("bigquery")
     .option("table", preds_table)
     .option("writeMethod", "direct")
     .mode("overwrite")
     .save())
    print(f"Predictions written to {preds_table}")

    spark.stop()

if __name__ == "__main__":
    main()
'''

print(f'PySpark script defined: {len(SPARK_SCRIPT)} characters')
````

### Upload to GCS

````python
bucket = gcs.bucket(GCS_BUCKET)
blob = bucket.blob('data_ai_overview/spark/fraud_detection.py')
blob.upload_from_string(SPARK_SCRIPT)

script_uri = f'gs://{GCS_BUCKET}/data_ai_overview/spark/fraud_detection.py'
print(f'PySpark script uploaded to {script_uri}')
````

### Submit the Batch Job

The [`BatchControllerClient`](https://cloud.google.com/python/docs/reference/dataproc/latest/google.cloud.dataproc_v1.services.batch_controller.BatchControllerClient) submits and monitors Dataproc Serverless batch jobs.

**Executor configuration** for batch jobs uses the same Spark properties as interactive sessions. Key settings:

| Property | Value | Effect |
|---|---|---|
| `spark.executor.instances` | `4` | 4 executor JVMs |
| `spark.executor.cores` | `4` | 4 CPU cores per executor |
| `spark.executor.memory` | `4g` | 4 GB RAM per executor |
| `spark.driver.memory` | `4g` | 4 GB RAM for the driver |
| `spark.dynamicAllocation.enabled` | `false` | Fixed executor count (required for XGBoost barrier mode) |

**Why disable dynamic allocation?** XGBoost's `SparkXGBClassifier` uses [barrier execution mode](https://spark.apache.org/docs/latest/barrier-mode.html), which requires all executors to be present simultaneously. Dynamic allocation can scale executors down mid-job, breaking barrier tasks. For jobs using only MLlib models (no XGBoost), you can re-enable dynamic allocation with `minExecutors`/`maxExecutors`.

````python
batch_client = dataproc_v1.BatchControllerClient(
    client_options={'api_endpoint': f'{REGION}-dataproc.googleapis.com:443'}
)

batch_id = f'fraud-detection-{int(time.time())}'

batch = dataproc_v1.Batch()
batch.pyspark_batch = dataproc_v1.PySparkBatch(
    main_python_file_uri=script_uri,
    args=[PROJECT_ID, BQ_DATASET, BQ_SOURCE],
)
batch.runtime_config = dataproc_v1.RuntimeConfig(version='2.2')

# Executor and driver configuration
# Dynamic allocation is disabled because XGBoost's SparkXGBClassifier
# uses barrier execution mode, which requires a fixed executor count.
batch.runtime_config.properties.update({
    'spark.executor.instances': '4',
    'spark.executor.cores': '4',
    'spark.executor.memory': '4g',
    'spark.driver.memory': '4g',
    'spark.dynamicAllocation.enabled': 'false',
})
batch.environment_config = dataproc_v1.EnvironmentConfig(
    execution_config=dataproc_v1.ExecutionConfig(
        staging_bucket=GCS_BUCKET,
    )
)

print(f'Submitting batch job: {batch_id}')
print(f'  Script:    {script_uri}')
print(f'  Executors: 4 × 4 cores × 4 GB (fixed — XGBoost requires barrier mode)')
print(f'  Args:      [{PROJECT_ID}, {BQ_DATASET}, {BQ_SOURCE}]')
t0 = time.time()

operation = batch_client.create_batch(
    request=dataproc_v1.CreateBatchRequest(
        parent=f'projects/{PROJECT_ID}/locations/{REGION}',
        batch=batch,
        batch_id=batch_id,
    )
)

result = operation.result()
elapsed = time.time() - t0
print(f'\nBatch job completed in {elapsed:.0f}s ({elapsed/60:.1f} min)')
print(f'State: {dataproc_v1.Batch.State(result.state).name}')
````

### Review Results

The batch job wrote two tables to BigQuery:
- `spark_fraud_metrics` — AUC-ROC, AUC-PR, and F1 for all three models (including pip-installed XGBoost)
- `spark_fraud_predictions` — predictions from the best model on the test set

## Custom Containers — Bring Your Own Environment

Dataproc Serverless runs your code on [managed Spark containers](https://cloud.google.com/dataproc-serverless/docs/concepts/custom-containers). By default, these include PySpark, the BigQuery connector, and common libraries. But what if you need:

- **Additional Python packages** — scikit-learn, XGBoost, specialized ML libraries
- **R packages** for SparkR workloads — dplyr, caret, tidyverse
- **System libraries** — C/C++ shared libraries for proprietary code
- **Proprietary libraries** — internal packages, custom JARs, licensed software

Custom containers solve this. You build a Docker image that extends the Dataproc base image, add whatever you need, push it to Artifact Registry, and reference it when submitting your batch job.

This is a **key capability** — it means Dataproc Serverless is not limited to the pre-installed package set. Any software that runs in a container can run on serverless Spark.

### The Dockerfile

The Dockerfile extends the official [Dataproc Serverless base image](https://cloud.google.com/dataproc-serverless/docs/concepts/custom-containers#base_image) and adds your dependencies. The base image already includes PySpark, Hadoop, and the BigQuery connector — you just layer on what you need.

**Important:** The base image tag must match the runtime version you specify in the batch job submission (e.g., `2.2-debian12` for runtime `2.2`).

````python
CUSTOM_DOCKERFILE = '''
# Extend the Dataproc Serverless base image
# Tag must match the runtime version used in batch job submission
FROM us-docker.pkg.dev/cloud-dataproc/spark/spark:2.2-debian12

# Switch to root to install system packages
USER root

# --- System libraries ---
# Add any C/C++ libraries needed by Python packages or proprietary code
RUN apt-get update && apt-get install -y --no-install-recommends \\
    libgomp1 \\
    liblapack-dev \\
    && rm -rf /var/lib/apt/lists/*

# --- Python packages ---
# Add ML libraries, data processing tools, or proprietary packages
RUN pip install --no-cache-dir \\
    pandas==2.2.* \\
    scikit-learn==1.5.* \\
    xgboost==2.1.* \\
    shap==0.46.* \\
    matplotlib==3.9.*

# --- R packages (for SparkR workloads) ---
# Install R packages that will be available to SparkR UDFs
RUN R -e "install.packages(c(\'dplyr\', \'caret\', \'glmnet\'), repos=\'https://cran.rstudio.com/\')"

# --- Proprietary / internal packages ---
# Example: copy a private wheel and install it
# COPY my_internal_lib-1.0.0-py3-none-any.whl /tmp/
# RUN pip install /tmp/my_internal_lib-1.0.0-py3-none-any.whl

# Switch back to the default non-root user
USER spark
'''

print('Custom Dockerfile for Dataproc Serverless:')
print(CUSTOM_DOCKERFILE)
````

### Building and Pushing the Image

Build the Docker image and push it to [Artifact Registry](https://cloud.google.com/artifact-registry/docs/overview). This is typically done once (or via CI/CD), not in the notebook itself.

```bash
# Create an Artifact Registry repository (one-time setup)
gcloud artifacts repositories create dataproc-custom \
    --repository-format=docker \
    --location=us-central1 \
    --description="Custom Dataproc Serverless images"

# Build and push
IMAGE_URI="us-central1-docker.pkg.dev/${PROJECT_ID}/dataproc-custom/spark-ml:latest"

docker build -t ${IMAGE_URI} .
docker push ${IMAGE_URI}
```

Or use [Cloud Build](https://cloud.google.com/build/docs/overview) for a fully managed build:

```bash
gcloud builds submit --tag ${IMAGE_URI} .
```

### Using the Custom Container in a Batch Job

Reference the custom container image via `runtime_config.container_image` when submitting a batch job. Everything else works exactly the same — the only difference is that your script now runs in an environment with your additional packages.

````python
CUSTOM_IMAGE_URI = f'{REGION}-docker.pkg.dev/{PROJECT_ID}/dataproc-custom/spark-ml:latest'

# Build the batch job configuration with the custom container
custom_batch = dataproc_v1.Batch()
custom_batch.pyspark_batch = dataproc_v1.PySparkBatch(
    main_python_file_uri=script_uri,
    args=[PROJECT_ID, BQ_DATASET, BQ_SOURCE],
)
custom_batch.runtime_config = dataproc_v1.RuntimeConfig(
    version='2.2',
    container_image=CUSTOM_IMAGE_URI,  # <-- Reference the custom container
)
custom_batch.environment_config = dataproc_v1.EnvironmentConfig(
    execution_config=dataproc_v1.ExecutionConfig(
        staging_bucket=GCS_BUCKET,
    )
)

print('Batch job configuration with custom container:')
print(f'  Container image: {CUSTOM_IMAGE_URI}')
print(f'  Script:          {script_uri}')
print(f'  Runtime:         2.2')
print()
print('To submit this job:')
print('  operation = batch_client.create_batch(')
print(f'      parent="projects/{PROJECT_ID}/locations/{REGION}",')
print(f'      batch=custom_batch,')
print(f'      batch_id="fraud-custom-{int(time.time())}"')
print('  )')
print()
print('The script now has access to scikit-learn, XGBoost, SHAP, and all other')
print('packages installed in the custom container.')
````

### What Custom Containers Enable

With custom containers, your PySpark scripts can use any library — not just what is pre-installed. Common patterns:

| Use Case | Packages Added | Why |
|---|---|---|
| Advanced ML | `scikit-learn`, `xgboost`, `lightgbm` | Ensemble methods, preprocessing pipelines |
| Model explainability | `shap`, `lime` | Feature importance beyond Spark's built-in metrics |
| Statistical analysis | `scipy`, `statsmodels` | Hypothesis testing, time series decomposition |
| SparkR data science | `dplyr`, `caret`, `tidyverse` | R ecosystem inside Spark |
| Proprietary code | Internal wheels, licensed JARs | Organization-specific logic |
| System libraries | `libgomp`, `liblapack`, `libcurl` | C/C++ dependencies for compiled packages |

The container is built once and reused across jobs. When your dependency set changes, build a new version and update the image tag.

## SparkR on Dataproc Serverless

Dataproc Serverless supports [SparkR](https://spark.apache.org/docs/latest/sparkr.html) alongside PySpark — R users get familiar syntax backed by Spark's distributed engine.

**Interactive vs batch:** The `DataprocSparkSession` used above connects via [Spark Connect](https://spark.apache.org/docs/latest/spark-connect-overview.html), which is Python-only. For SparkR, use **batch jobs**: write an `.R` script, upload to GCS, and submit via `SparkRBatch`. Same infrastructure, same auto-scaling, different language.

To add R packages beyond the pre-installed set (e.g., `dplyr`, `caret`, `tidyverse`), use a custom container as shown in the previous section.

### The SparkR Script

This script mirrors the PySpark workflow — same dataset, same fraud detection task — but in R. It uses SparkR's `spark.glm()` to train a logistic regression (binomial GLM), predicts on the test set, and writes results to BigQuery for comparison with the PySpark models.

````python
SPARKR_SCRIPT = '''#!/usr/bin/env Rscript
# SparkR fraud detection — logistic regression on Dataproc Serverless Batch
library(SparkR)

args <- commandArgs(trailingOnly = TRUE)
project_id <- args[1]
bq_dataset <- args[2]
bq_source  <- args[3]

sparkR.session(appName = "fraud-detection-sparkr")

# --- Read from BigQuery ---
df <- read.df(source = "bigquery", table = bq_source)
cat(sprintf("Loaded %d rows, %d columns\n", nrow(df), ncol(df)))

# --- Train/test split ---
splits <- randomSplit(df, weights = c(0.8, 0.2), seed = 42)
train <- splits[[1]]
test  <- splits[[2]]
cat(sprintf("Train: %d rows, Test: %d rows\n", nrow(train), nrow(test)))

# --- Logistic regression with SparkR GLM ---
cat("Training SparkR GLM (logistic regression)...\n")
model <- spark.glm(
    train,
    Class ~ V1 + V2 + V3 + V4 + V5 + V6 + V7 + V8 + V9 + V10 +
            V11 + V12 + V13 + V14 + V15 + V16 + V17 + V18 + V19 + V20 +
            V21 + V22 + V23 + V24 + V25 + V26 + V27 + V28 + Amount,
    family = "binomial"
)
cat("Training complete.\n")
cat(sprintf("Null deviance: %.1f on %d DF\n",
    summary(model)$null.deviance, summary(model)$df.null))
cat(sprintf("Residual deviance: %.1f on %d DF\n",
    summary(model)$deviance, summary(model)$df.residual))
cat(sprintf("AIC: %.1f\n", summary(model)$aic))

# --- Predict on test set ---
# For binomial GLM, the prediction column is the probability (response scale)
predictions <- predict(model, test)
predictions <- withColumn(predictions, "predicted_class",
    ifelse(predictions$prediction > 0.5, 1, 0))

# --- Confusion matrix ---
cm <- collect(count(groupBy(
    select(predictions, "Class", "predicted_class"),
    "Class", "predicted_class"
)))
cm <- cm[order(cm$Class, cm$predicted_class), ]
cat("\nConfusion matrix:\n")
print(cm)

total_count   <- sum(cm$count)
correct_count <- sum(cm$count[cm$Class == cm$predicted_class])
cat(sprintf("Accuracy: %d/%d (%.4f)\n", correct_count, total_count,
    correct_count / total_count))

# --- Write predictions to BigQuery ---
pred_out   <- select(predictions, "Time", "Amount", "Class",
                     "predicted_class", "prediction")
pred_table <- paste0(project_id, ".", bq_dataset, ".sparkr_fraud_predictions")
write.df(pred_out, source = "bigquery", table = pred_table,
         writeMethod = "direct", mode = "overwrite")
cat(sprintf("Predictions written to %s\n", pred_table))

sparkR.session.stop()
cat("Done.\n")
'''

print(f'SparkR script defined: {len(SPARKR_SCRIPT)} characters')
````

### Submit SparkR Batch Job

Upload the R script to GCS and submit via `SparkRBatch` — the same pattern as PySpark batch, with `spark_r_batch` instead of `pyspark_batch`.

````python
blob = bucket.blob('data_ai_overview/spark/fraud_detection.R')
blob.upload_from_string(SPARKR_SCRIPT)
sparkr_uri = f'gs://{GCS_BUCKET}/data_ai_overview/spark/fraud_detection.R'
print(f'SparkR script uploaded to {sparkr_uri}')

sparkr_batch_id = f'fraud-sparkr-{int(time.time())}'

sparkr_batch = dataproc_v1.Batch()
sparkr_batch.spark_r_batch = dataproc_v1.SparkRBatch(
    main_r_file_uri=sparkr_uri,
    args=[PROJECT_ID, BQ_DATASET, BQ_SOURCE],
)
sparkr_batch.runtime_config = dataproc_v1.RuntimeConfig(version='2.2')
sparkr_batch.environment_config = dataproc_v1.EnvironmentConfig(
    execution_config=dataproc_v1.ExecutionConfig(staging_bucket=GCS_BUCKET)
)

print(f'Submitting SparkR batch job: {sparkr_batch_id}')
t0 = time.time()

operation = batch_client.create_batch(
    request=dataproc_v1.CreateBatchRequest(
        parent=f'projects/{PROJECT_ID}/locations/{REGION}',
        batch=sparkr_batch,
        batch_id=sparkr_batch_id,
    )
)

result = operation.result()
elapsed = time.time() - t0
print(f'\nSparkR batch completed in {elapsed:.0f}s ({elapsed/60:.1f} min)')
print(f'State: {dataproc_v1.Batch.State(result.state).name}')
````

### Review SparkR Results

The SparkR batch job wrote predictions to BigQuery. Read them back and compute evaluation metrics — showing the natural cross-language interop that BigQuery enables: R trains, Python evaluates.

````python
sparkr_preds = bq.query(f'''
SELECT
    CAST(Class AS INT64) AS actual,
    CAST(predicted_class AS INT64) AS predicted,
    COUNT(*) AS count
FROM `{PROJECT_ID}.{BQ_DATASET}.sparkr_fraud_predictions`
GROUP BY actual, predicted
ORDER BY actual, predicted
''').to_dataframe()

print('SparkR GLM — Confusion matrix (test set):')
print(sparkr_preds.to_string(index=False))

total = sparkr_preds['count'].sum()
correct = sparkr_preds[sparkr_preds['actual'] == sparkr_preds['predicted']]['count'].sum()
print(f'\nAccuracy: {correct}/{total} ({correct/total:.4f})')
````

## The Migration Story — From Persistent Clusters to Serverless

Traditional Hadoop/Spark deployments require persistent clusters — machines that run 24/7, whether you are using them or not. This model carries significant operational overhead:

| Concern | Persistent Cluster | Dataproc Serverless |
|---|---|---|
| **Provisioning** | Choose machine types, node counts, configure networking | None — infrastructure auto-provisions |
| **Scaling** | Manual or autoscaling policies to configure | Automatic — scales with the workload |
| **Idle cost** | Paying for running machines even when no jobs are active | Zero cost when idle — pay only for job runtime |
| **Maintenance** | OS patches, Spark upgrades, HDFS health | Fully managed — Google handles all maintenance |
| **Security** | Configure firewalls, Kerberos, key management | VPC-native, IAM-based, no cluster-level config |
| **Multi-tenancy** | Shared cluster contention, YARN queue management | Each job gets isolated resources |

**The shift:**
- **Before:** Maintain a long-running Dataproc cluster (or on-prem Hadoop cluster). Submit jobs to it. Worry about sizing, utilization, and upgrades.
- **After:** Write your Spark code. Submit it to Dataproc Serverless. Infrastructure appears, runs the job, disappears. You manage code, not clusters.

This is the same pattern as BigQuery (serverless SQL) and Cloud Run (serverless containers) — you bring the logic, Google brings the infrastructure. Dataproc Serverless extends this to the full Spark/SparkR ecosystem.

**When you still need Dataproc on Compute Engine (persistent clusters):**
- Workloads that need HDFS for persistent storage
- Long-running streaming jobs (though Serverless supports streaming too)
- Highly customized Hadoop ecosystem components (HBase, Hive Metastore with specific configs)
- Jobs that require GPU-attached workers (not yet available in Serverless)
