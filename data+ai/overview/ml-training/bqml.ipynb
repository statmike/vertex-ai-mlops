# BigQuery ML — Train Models Entirely in SQL

This notebook demonstrates [BigQuery ML (BQML)](https://cloud.google.com/bigquery/docs/bqml-introduction) — training, evaluating, and predicting with ML models using only SQL, without moving data out of BigQuery.

BQML eliminates the extract-transform-load-train cycle: data stays in BigQuery, training runs on BigQuery infrastructure, and predictions are SQL queries. For data analysts who already know SQL, this is the fastest path from data to model.

**What it creates:**
- `fraud_xgb` — a BQML `BOOSTED_TREE_CLASSIFIER` model with Vizier hyperparameter tuning
- Evaluation metrics, batch predictions, and feature importance — all via SQL

**The story:** The sibling `bigframes-local.ipynb` notebook trains XGBoost locally after pulling data to the client. This notebook does the same thing — but the data never leaves BigQuery, hyperparameter tuning is automatic, and the entire workflow is SQL. Beyond BQML's built-in model types, BigQuery also supports remote models (Vertex AI endpoints), remote functions (Cloud Run), and native Python UDFs — all queryable with standard SQL.

**Dataset:** [`bigquery-public-data.ml_datasets.ulb_fraud_detection`](https://console.cloud.google.com/bigquery?p=bigquery-public-data&d=ml_datasets&t=ulb_fraud_detection) — 284,807 credit card transactions with 492 frauds (0.17%). Features `V1`–`V28` are PCA-transformed, plus `Time`, `Amount`, and binary `Class` (0=legitimate, 1=fraud).

## CREATE MODEL — Boosted Tree Classifier with Vizier Tuning

BQML's [`BOOSTED_TREE_CLASSIFIER`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create-boosted-tree) trains a gradient-boosted tree directly on a BigQuery table. The data never leaves BigQuery — no export, no local memory, no separate training infrastructure.

Setting `num_trials` activates [Vertex AI Vizier](https://cloud.google.com/vertex-ai/docs/vizier/overview) for automatic hyperparameter tuning. Vizier explores the search space intelligently (Bayesian optimization), trying different combinations of `learn_rate`, `max_tree_depth`, `subsample`, and other parameters across the specified number of trials.

**Key parameters:**

| Parameter | Value | Purpose |
|---|---|---|
| `model_type` | `BOOSTED_TREE_CLASSIFIER` | XGBoost-based gradient boosting |
| `input_label_cols` | `['Class']` | Target column (0=legitimate, 1=fraud) |
| `auto_class_weights` | `TRUE` | Compensate for 578:1 class imbalance |
| `num_trials` | `10` | Run 10 Vizier hyperparameter tuning trials |
| `max_parallel_trials` | `5` | Run up to 5 trials concurrently |
| `enable_global_explain` | `TRUE` | Compute feature attributions after training |

````python
start = time.time()

query = f"""
CREATE OR REPLACE MODEL `{PROJECT_ID}.{BQ_DATASET}.fraud_xgb`
OPTIONS(
    model_type = 'BOOSTED_TREE_CLASSIFIER',
    input_label_cols = ['Class'],
    auto_class_weights = TRUE,
    num_trials = 10,
    max_parallel_trials = 5,
    enable_global_explain = TRUE
) AS
SELECT * EXCEPT(Time)
FROM `{BQ_SOURCE}`
"""

bq.query(query).result()
elapsed = time.time() - start
print(f'Model trained in {elapsed:.0f}s ({elapsed/60:.1f} min)')
````

Check which hyperparameters Vizier selected across the 10 trials.

````python
trials_query = f"""
SELECT
    trial_id,
    status,
    is_optimal,
    eval_loss,
    hparam_tuning_evaluation_metrics.roc_auc AS roc_auc,
    hyperparameters.learn_rate,
    hyperparameters.max_tree_depth,
    hyperparameters.subsample,
    hyperparameters.l1_reg,
    hyperparameters.l2_reg
FROM ML.TRIAL_INFO(MODEL `{PROJECT_ID}.{BQ_DATASET}.fraud_xgb`)
ORDER BY trial_id
"""
trials_df = bq.query(trials_query).to_dataframe()
print(f'{len(trials_df)} trial(s) completed:')
trials_df
````

## ML.EVALUATE — Model Evaluation

[`ML.EVALUATE`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-evaluate) returns standard classification metrics. By default it evaluates on the held-out data split that BQML reserved during training.

````python
eval_query = f"""
SELECT *
FROM ML.EVALUATE(MODEL `{PROJECT_ID}.{BQ_DATASET}.fraud_xgb`)
"""
eval_df = bq.query(eval_query).to_dataframe()
eval_df
````

````python
roc_query = f"""
SELECT *
FROM ML.ROC_CURVE(MODEL `{PROJECT_ID}.{BQ_DATASET}.fraud_xgb`)
"""
roc_df = bq.query(roc_query).to_dataframe()
roc_df = roc_df.sort_values('false_positive_rate')

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(roc_df['false_positive_rate'], roc_df['recall'], color='#4285F4', lw=2, label='BQML Boosted Tree')
ax.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--', label='Random')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate (Recall)')
ax.set_title('ROC Curve — ML.ROC_CURVE')
ax.legend(loc='lower right')
ax.set_xlim([0, 1])
ax.set_ylim([0, 1.05])
plt.tight_layout()
plt.show()
````

---
## ML.PREDICT — Batch Predictions

[`ML.PREDICT`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-predict) scores new data using the trained model. The input is any query that produces the same feature columns — it could be another table, a view, or a subquery. Predictions run at BigQuery scale without moving data.

````python
predict_query = f"""
SELECT
    predicted_Class,
    predicted_Class_probs,
    Class AS actual_class,
    Amount
FROM ML.PREDICT(
    MODEL `{PROJECT_ID}.{BQ_DATASET}.fraud_xgb`,
    (SELECT * EXCEPT(Time) FROM `{BQ_SOURCE}` LIMIT 1000)
)
"""
pred_df = bq.query(predict_query).to_dataframe()
print(f'Predictions: {len(pred_df)} rows')
pred_df.head(10)
````

````python
correct = (pred_df['predicted_Class'] == pred_df['actual_class']).sum()
total = len(pred_df)
print(f'Accuracy on sample: {correct}/{total} ({correct/total:.1%})')
print(f'\nPrediction distribution:')
print(pred_df['predicted_Class'].value_counts().to_string())
````

## ML.EXPLAIN_PREDICT — Feature Importance

[`ML.EXPLAIN_PREDICT`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-explain-predict) returns per-row feature attributions alongside predictions. Because we trained with `enable_global_explain = TRUE`, we can also retrieve global feature importance with [`ML.GLOBAL_EXPLAIN`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-global-explain).

````python
global_explain_query = f"""
SELECT *
FROM ML.GLOBAL_EXPLAIN(MODEL `{PROJECT_ID}.{BQ_DATASET}.fraud_xgb`)
ORDER BY attribution DESC
"""
explain_df = bq.query(global_explain_query).to_dataframe()
print('Global feature importance:')
explain_df
````

````python
top_n = min(15, len(explain_df))
top = explain_df.head(top_n).iloc[::-1]

fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(top['feature'], top['attribution'], color='#4285F4')
ax.set_xlabel('Attribution')
ax.set_title(f'Top {top_n} Feature Importances — ML.GLOBAL_EXPLAIN')
plt.tight_layout()
plt.show()
````

## Remote Models — Vertex AI Endpoints in SQL

BQML isn't limited to its built-in model types. [Remote models](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create-remote-model) let you register a Vertex AI Endpoint (or any model deployed behind one) and query it with `ML.PREDICT` as if it were a native BQML model.

This bridges the gap between custom-trained models (PyTorch, TensorFlow, scikit-learn deployed on Vertex AI) and SQL-based analytics. Data stays in BigQuery — predictions are sent to the endpoint and results flow back as query results.

**Prerequisites:**
- A [BigQuery Cloud resource connection](https://cloud.google.com/bigquery/docs/create-cloud-resource-connection) with the service account granted `Vertex AI User` role
- A deployed Vertex AI Endpoint

**The SQL pattern:**

```sql
-- Step 1: Create a Cloud resource connection (one-time setup)
-- Done in the Console or via bq mk --connection

-- Step 2: Register the remote model
CREATE OR REPLACE MODEL `project.dataset.my_remote_model`
  REMOTE WITH CONNECTION `project.region.connection_name`
  OPTIONS(
    endpoint = 'https://region-aiplatform.googleapis.com/v1/projects/project/locations/region/endpoints/ENDPOINT_ID'
  );

-- Step 3: Query it like any BQML model
SELECT *
FROM ML.PREDICT(
  MODEL `project.dataset.my_remote_model`,
  (SELECT feature_1, feature_2 FROM `project.dataset.input_table`)
);
```

The connection's service account handles authentication to the endpoint. From the analyst's perspective, it's just another `ML.PREDICT` call — no Python, no API clients, no data movement.

## Remote Functions — Custom Logic via Cloud Run

[Remote functions](https://cloud.google.com/bigquery/docs/reference/standard-sql/remote-functions) let you expose arbitrary code running on Cloud Run (or Cloud Functions) as a SQL-callable function. Any logic that can be served as an HTTP endpoint — custom transformations, third-party API calls, specialized scoring — becomes a function you can use in `SELECT`, `WHERE`, or `JOIN` clauses.

**Prerequisites:**
- A Cloud Run service (or Cloud Function) that accepts JSON input and returns JSON output
- A BigQuery Cloud resource connection

**The SQL pattern:**

```sql
-- Create a remote function pointing to a Cloud Run service
CREATE OR REPLACE FUNCTION `project.dataset.my_remote_function`(input_text STRING)
RETURNS STRING
REMOTE WITH CONNECTION `project.region.connection_name`
OPTIONS(
  endpoint = 'https://my-service-abc123.run.app',
  max_batching_rows = 1000
);

-- Use it in any query
SELECT
  input_text,
  `project.dataset.my_remote_function`(input_text) AS result
FROM `project.dataset.my_table`;
```

BigQuery batches rows and sends them to the endpoint in parallel, handling retries and scaling automatically. The function appears in the schema like any other UDF — consumers don't need to know it calls an external service.

## Native Python UDFs — PyPI Packages in SQL

[Native Python UDFs](https://cloud.google.com/bigquery/docs/python-user-defined-functions) let data scientists write pure Python directly in SQL — with full access to PyPI packages. The Python code runs inside BigQuery's infrastructure on a secure sandbox, so data never leaves BigQuery.

This is particularly powerful for transformations that are awkward in SQL but trivial in Python: string parsing, statistical functions, custom feature engineering, or anything available in a PyPI package.

**Key feature:** The `packages` option lists PyPI dependencies that BigQuery automatically installs. No Docker containers, no Cloud Run services — just declare the package and use it.

````python
udf_query = f"""
CREATE OR REPLACE FUNCTION `{PROJECT_ID}.{BQ_DATASET}.zscore_amount`(val FLOAT64, mean_val FLOAT64, std_val FLOAT64)
RETURNS FLOAT64
LANGUAGE python
OPTIONS(
    runtime_version = 'python-3.11',
    packages = ['numpy'],
    entry_point = 'zscore'
)
AS r\"\"\"
import numpy as np

def zscore(val, mean_val, std_val):
    if std_val == 0:
        return 0.0
    return float(np.float64((val - mean_val) / std_val))
\"\"\";
"""
bq.query(udf_query).result()
print('Python UDF created')
````

````python
use_udf_query = f"""
WITH stats AS (
    SELECT AVG(Amount) AS mean_amt, STDDEV(Amount) AS std_amt
    FROM `{BQ_SOURCE}`
)
SELECT
    Amount,
    `{PROJECT_ID}.{BQ_DATASET}.zscore_amount`(Amount, stats.mean_amt, stats.std_amt) AS amount_zscore,
    Class
FROM `{BQ_SOURCE}`, stats
ORDER BY amount_zscore DESC
LIMIT 10
"""
udf_df = bq.query(use_udf_query).to_dataframe()
print('Transactions with highest z-score amounts (using Python UDF):')
udf_df
````

The UDF above is deliberately simple to show the pattern. In practice, Python UDFs shine when you need:
- **Complex string parsing** with `regex` or domain-specific libraries
- **Statistical computations** with `scipy` or `statsmodels`
- **Feature engineering** with `scikit-learn` preprocessors
- **Custom scoring** with pre-trained model weights loaded from GCS

Each of these is a `packages = [...]` declaration away.
