# R - Notebook Workflow

Runtime: **interactive R kernel**.  Develop and run the shared GLM workflow end-to-end right in the notebook - prepare data in BigQuery, read it into R, fit the model, and evaluate it.  This is the starting point; the other notebooks in the series take *this same workflow* and run it as managed jobs.

---
Part of the **R** series - see the [series guide](./readme.md). This series runs *one shared workflow* (a logistic-regression fraud-detection GLM on `bigquery-public-data.ml_datasets.ulb_fraud_detection`) across multiple Google Cloud runtimes so you can compare them for your own work.

---
**The shared data & workflow**

Every notebook in this series uses the same task: a logistic regression (`glm`, `family = binomial`) that classifies credit-card transactions as fraudulent (`Class = 1`) or normal (`Class = 0`), from `bigquery-public-data.ml_datasets.ulb_fraud_detection` (284,807 rows; PCA features `V1..V28` plus `Time` and `Amount`).

The data is prepared once into a `*_prepped` table that adds `transaction_id` and a `splits` column (`TRAIN`/`VALIDATE`/`TEST`).  Reads use the modern, columnar path - the **BigQuery Storage Read API** (Arrow) via `bigrquerystorage`, or `arrow` over Parquet - never a CSV export.  See [R - Reading BigQuery Iceberg Tables](./R%20-%20Reading%20BigQuery%20Iceberg%20Tables.ipynb) for the data-access deep dive.

**Prerequisites**
- An R environment (Vertex AI Workbench Instance with an **R (Local)** kernel - see the series [readme](./readme.md)).
- R packages: `bigrquery`, `bigrquerystorage`, `DBI`, `dplyr`.  In a Workbench conda `r` env: `conda install -c conda-forge r-bigrquery r-bigrquerystorage r-dbi r-dplyr`.

---
## Setup

inputs:

```r
project_id <- 'statmike-mlops-349915'
project_id
```

```r
region     <- 'us-central1'
series     <- 'r'
experiment <- 'bigquery-data'

# BigQuery parameters
bq_project <- project_id
bq_dataset <- series
bq_table   <- paste0(experiment, '_prepped')   # prepared table with transaction_id + splits
bq_region  <- substr(region, 1, 2)
bq_source  <- 'bigquery-public-data.ml_datasets.ulb_fraud_detection'

# key columns
var_target <- 'Class'
var_omit   <- c('transaction_id', 'splits')
```

packages:

```r
# install any missing packages (idempotent; needed on Colab / fresh environments)
required <- c("bigrquery", "bigrquerystorage", "dplyr")
missing <- required[!sapply(required, requireNamespace, quietly = TRUE)]
if (length(missing)) install.packages(missing, repos = "https://cloud.r-project.org")

library(bigrquery)
library(bigrquerystorage)  # enables the fast Arrow / Storage Read API download path
library(dplyr)
```

---
## Prepare Data For Models

R is great for data prep, but processing done in R must be reproduced at serving time on new data.  To keep that boundary clean, this step does the prep **in BigQuery** (add a row id and assign splits) and lands a `*_prepped` table.  R then just reads the result.

Create the dataset if it does not already exist:

```r
bq_ds <- bq_dataset(bq_project, bq_dataset)
if (!bq_dataset_exists(bq_ds)) {
    bq_dataset_create(bq_ds, location = bq_region)
}
bq_dataset_exists(bq_ds)
```

Create the prepared table - add `transaction_id` and assign `TRAIN`/`VALIDATE`/`TEST` splits deterministically:

```r
prep_sql <- sprintf('
CREATE TABLE IF NOT EXISTS `%s.%s.%s` AS
WITH add_id AS (SELECT *, GENERATE_UUID() AS transaction_id FROM `%s`)
SELECT *,
    CASE
        WHEN MOD(ABS(FARM_FINGERPRINT(transaction_id)), 10) < 8 THEN "TRAIN"
        WHEN MOD(ABS(FARM_FINGERPRINT(transaction_id)), 10) < 9 THEN "VALIDATE"
        ELSE "TEST"
    END AS splits
FROM add_id
', bq_project, bq_dataset, bq_table, bq_source)
cat(prep_sql)
```

```r
prep <- bq_perform_query(prep_sql, billing = bq_project)
bq_job_wait(prep)
```

### Read The Splits Into R

Read each split with column projection (`EXCEPT`) and a row filter pushed to BigQuery.  With `bigrquerystorage` installed, `bq_table_download()` uses the Storage Read API (Arrow) automatically - fast and parallel, no GCS staging.

```r
get_data <- function(s) {
    query <- sprintf('
        SELECT * EXCEPT(%s)
        FROM `%s.%s.%s`
        WHERE splits = "%s"
    ', paste(var_omit, collapse = ','), bq_project, bq_dataset, bq_table, s)
    bq_table_download(bq_project_query(bq_project, query), n_max = Inf)
}

train <- get_data("TRAIN")
test  <- get_data("TEST")
dim(train); dim(test)
```

```r
head(train, 2)
```

---
## Train Model

Fit logistic regression with `glm`:

```r
model <- glm(
    as.formula(paste0(var_target, " ~ .")),
    data = train,
    family = binomial
)
summary(model)
```

---
## Evaluate Model

Score the held-out `TEST` split and compare predictions to actuals as a confusion matrix:

```r
preds  <- predict(model, test, type = "response")
actual <- test[, var_target]
names(actual) <- 'actual'
results <- cbind(actual, data.frame(pred = round(preds)))
table(results)
```

---
## Save The Model

Persist the fitted model.  The job-based runtimes in this series save to `AIP_MODEL_DIR` in GCS; here we just write locally.

> **A note on R model size.**  An R `glm` object stores a full copy of the training data inside it (`model$model`, `residuals`, `fitted.values`, `effects`, the response vector, etc.), so the saved `.rds` can be surprisingly large — here ~164 MB even though the model is really just a vector of coefficients.  This matters most at **serving** time, where the `.rds` is baked into a container image and loaded into memory ([R - Serving Predictions](./R%20-%20Serving%20Predictions.ipynb)).
>
> Dropping those fit-time data components keeps point predictions **identical** — `predict(..., type = "response")` only needs the coefficients, `terms`, `family`, and `qr` — while shrinking the object substantially (~164 MB → ~55 MB here).  The optional cell below does the trim.  Keep the full object instead if you need `se.fit`, residual diagnostics, or to `update()` / refit the model.

```r
saveRDS(model, "model.rds")
```

**Optional — trim the model before saving for serving.**  Removes the fit-time data copies; point predictions are unchanged.

```r
# strip fit-time data the GLM carries but point-prediction does not need
# (predict.glm with type="response" needs coefficients, terms, family, and qr)
trim_glm <- function(m) {
    m$model             <- NULL  # the model frame = full copy of the training data
    m$data              <- NULL
    m$residuals         <- NULL
    m$fitted.values     <- NULL
    m$effects           <- NULL
    m$linear.predictors <- NULL
    m$weights           <- NULL
    m$prior.weights     <- NULL
    m$y                 <- NULL
    m
}

model_slim <- trim_glm(model)

# sanity check: predictions identical to the full model
stopifnot(all.equal(
    predict(model,      head(test, 100), type = "response"),
    predict(model_slim, head(test, 100), type = "response")
))

saveRDS(model_slim, "model.rds")
cat(sprintf("saved model.rds  (full: %.0f MB  ->  trimmed: %.0f MB)\n",
            as.numeric(object.size(model)) / 1e6,
            as.numeric(object.size(model_slim)) / 1e6))
```
