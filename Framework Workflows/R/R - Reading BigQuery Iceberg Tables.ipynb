# R - Reading BigQuery Iceberg Tables

Reading **large-scale** BigQuery data into **R** — a BigQuery **native** table and an **Apache Iceberg managed table** — across patterns from small to large, with a headline demonstration of **multithreaded, parallel Parquet reads** of the Iceberg table using [`arrow`](https://arrow.apache.org/docs/r/). No CSV staging, no [`duckdb`](https://duckdb.org/) middle layer — just BigQuery, Parquet, and R.

---
Part of the series of [**R**](./readme.md) workflows: running R code, training models with R, and serving predictions with R on Vertex AI and Google Cloud.

---

**⚠️ Prerequisite — run the setup notebook first**

This notebook only **reads** tables. Run the Python setup notebook first; it creates the **Apache Iceberg managed table** (from a large public dataset, so it shards into **many** Parquet files) and the BigLake connection:

➡️ [`data+ai/bq-iceberg/bq-iceberg-setup.ipynb`](../../data%2Bai/bq-iceberg/bq-iceberg-setup.ipynb)

An **[Apache Iceberg managed table](https://docs.cloud.google.com/bigquery/docs/biglake-iceberg-tables-in-bigquery)** (formerly *BigLake tables for Apache Iceberg in BigQuery*) stores its data as ordinary **Parquet** in *your own* GCS bucket — so R reads those files directly with `arrow`, **multithreaded across the many shards**.

**The Data:** `bigquery-public-data.chicago_taxi_trips.taxi_trips` — ~212M rows / ~83 GB. Large enough that the Iceberg table is many Parquet files, which is what makes the threading real. (The same table the [bq-reads](../../data%2Bai/tabular-data/bq-reads/readme.md) benchmark uses.)

**What this notebook shows**

| # | Pattern | Table | Package | Scale |
|---|---|---|---|---|
| A | Classic reads | native | `bigrquery`, `DBI`, `dplyr`+`dbplyr` | small (bounded) |
| B | Storage Read API | native | `bigrquery` + `bigrquerystorage` | large (bounded slice) |
| C | `arrow` on `/data` Parquet, **multithreaded** | Iceberg managed | `arrow` | large (all shards, parallel) |

> Reads here are **bounded** (aggregates / filters / limits). The full 212M-row table is far too large to pull into R memory — the point is *how* R reads at scale, in parallel, not materializing everything.

**Environment**

- An **R** kernel (Vertex AI Workbench **R (Local)**, or any R environment) — see the series [readme](./readme.md).
- R packages: `bigrquery`, `bigrquerystorage`, `DBI`, `dplyr`, `dbplyr`, `arrow`. In a Workbench conda `r` env: `conda install -c conda-forge r-bigrquery r-bigrquerystorage r-arrow r-dbi r-dplyr r-dbplyr`.

## Read Pattern A — Classic `bigrquery` Reads (native table, bounded)

Three idiomatic ways to pull from a **native** BigQuery table. Great for small/bounded results; here we cap or aggregate so nothing huge comes back.

**`bq_table_download()`** — a small `LIMIT` sample:

```r
tbl <- bq_project_query(project_id, sprintf('SELECT * FROM `%s` LIMIT 5', native_fqn))
ds_a1 <- bq_table_download(tbl, n_max = 5)
dim(ds_a1); head(ds_a1, 2)
```

**`DBI`** — a connection + `dbGetQuery()` with a bounded aggregate:

```r
con <- dbConnect(bigrquery::bigquery(), project = project_id, billing = project_id)
ds_a2 <- dbGetQuery(con, sprintf(
    'SELECT payment_type, COUNT(*) AS trips FROM `%s` GROUP BY payment_type ORDER BY trips DESC LIMIT 5',
    native_fqn))
ds_a2
```

**`dplyr` + `dbplyr`** — lazy table reference; verbs translate to SQL and run *in BigQuery*; `collect()` pulls only the small aggregate:

```r
# dbConnect: `project` is where the dataset LIVES (the public project), `billing` pays.
con_ds <- dbConnect(bigrquery::bigquery(),
                    project = "bigquery-public-data",   # taxi_trips lives here
                    dataset = "chicago_taxi_trips",
                    billing = project_id)               # your project is billed
ds_a3 <- tbl(con_ds, 'taxi_trips') %>%
    summarize(trips = n(), avg_fare = mean(fare, na.rm = TRUE)) %>%
    collect()
ds_a3
```

## Read Pattern B — Storage Read API (native table, large bounded slice)

With **`bigrquerystorage`** installed, `bq_table_download()` uses the **BigQuery Storage Read API** (Arrow) — fast, parallel, no GCS staging. We pull a **bounded slice** (a few columns, one time window) of the native table and time it.

```r
slice_sql <- sprintf("
    SELECT trip_seconds, trip_miles, fare, tips, trip_total, payment_type
    FROM `%s`
    WHERE trip_start_timestamp >= '2023-01-01' AND trip_start_timestamp < '2023-02-01'
", native_fqn)

t_b <- system.time({
    tbl <- bq_project_query(project_id, slice_sql)
    df_native <- bq_table_download(tbl, n_max = Inf)
})
cat(sprintf("Storage API read: %s rows x %s cols in %.1fs\n",
            format(nrow(df_native), big.mark=','), ncol(df_native), t_b['elapsed']))
head(df_native, 2)
```

## Read Pattern C — `arrow` on the Iceberg `/data` Parquet, Multithreaded (the headline)

The Iceberg managed table is stored as **many** Parquet shards under `<storage_uri>/data`. `arrow::open_dataset()` reads them directly from GCS as one lazy dataset and scans them **in parallel across its thread pool** — no BigQuery round-trip, no download step.

First, confirm there really are many files to parallelize across (the whole reason this demo needs a large table):

```r
data_uri <- sprintf('%s/data', storage_uri)
fs <- arrow::gs_bucket(gcs_bucket)
rel <- sub(sprintf('gs://%s/', gcs_bucket), '', data_uri)
files <- fs$GetFileInfo(arrow::FileSelector$create(rel, recursive = TRUE))
parquet_files <- Filter(function(f) grepl('\\.parquet$', f$path), files)

# arrow has TWO thread pools, both relevant to this read:
#   CPU pool (cpu_count)      - decompress + aggregate
#   I/O pool (io_thread_count) - fetch files from GCS
cat(sprintf(
    "Parquet shards: %d\nCPU threads (cpu_count): %d\nI/O threads (io_thread_count): %d\n",
    length(parquet_files), arrow::cpu_count(), arrow::io_thread_count()))
flush.console()
```

Open the shards as one lazy dataset (directly from GCS — no FUSE mount):

```r
ds <- arrow::open_dataset(data_uri, format = 'parquet')
ds
```

Lazy — inspect the schema without reading rows:

```r
ds$schema
```

**Aggregate across all shards, multithreaded.** `arrow` scans every Parquet file in parallel and returns only the grouped result — the full table never lands in R:

```r
ds %>%
    group_by(payment_type) %>%
    summarize(trips = n(), avg_total = mean(trip_total, na.rm = TRUE)) %>%
    arrange(desc(trips)) %>%
    collect()
```

### Multithreading & throughput

`arrow` parallelizes this read across **two** thread pools, and it matters which one is the bottleneck:

- **I/O pool** (`io_thread_count`) — fetches the Parquet shards from GCS concurrently. Reading ~1,300 files over the network is **I/O-bound**, so *this* pool is what drives throughput here.
- **CPU pool** (`cpu_count`) — decompresses and runs the aggregation once bytes are in memory.

Because the work is dominated by fetching many remote files, the **multi-file layout is the win**: arrow streams the shards in parallel rather than reading one giant file serially. Below we run one full parallel scan and report **throughput** (the honest metric) rather than a single-thread-vs-many race — with the I/O pool pinned to 1, a serial fetch of ~17 GB across ~1,300 files takes *many minutes*, which is itself the point.

```r
# one full parallel scan across all shards, reporting throughput (the honest metric
# for an I/O-bound read) rather than a misleading cpu-thread race.
gb_total <- sum(sapply(parquet_files, function(f) f$size)) / 1e9

scan_throughput <- function() {
    elapsed <- system.time({
        ds %>%
            group_by(payment_type) %>%
            summarize(trips = n(), avg_total = mean(trip_total, na.rm = TRUE)) %>%
            collect()
    })[['elapsed']]
    elapsed
}

elapsed <- scan_throughput()
cat(sprintf(
    "scanned %d shards / %.1f GB in %.1fs\n  throughput: %.2f GB/s, %.0f files/s\n",
    length(parquet_files), gb_total, elapsed, gb_total / elapsed, length(parquet_files) / elapsed))
flush.console()
```

#### Tuning the I/O pool

This read is **I/O-bound** — the time goes into fetching ~1,300 files from GCS, not CPU work — so the **I/O** thread pool (`io_thread_count`), not the CPU pool, is the relevant knob. Re-run the *same* scan with more I/O threads and compare throughput.

Whether it helps depends on your environment's **network/bandwidth headroom**: if the default pool already saturates the link, more threads won't move the number (you may well see ~1.0x). The lesson is the *method* — confirm which pool bounds the read, then size that pool — not that more threads is always faster.

> `arrow::set_io_thread_count()` resizes the pool used to read files; we restore the default afterward.

```r
default_io <- arrow::io_thread_count()

# baseline at the default I/O pool
arrow::set_io_thread_count(default_io)
t_default <- scan_throughput()

# more I/O threads for the same scan
more_io <- 32
arrow::set_io_thread_count(more_io)
t_more <- scan_throughput()

arrow::set_io_thread_count(default_io)   # restore

speedup <- t_default / t_more
verdict <- if (speedup > 1.15) {
    sprintf("more I/O threads helped (%.1fx) - the default pool was not saturating bandwidth", speedup)
} else {
    "~no change - the default I/O pool already saturates this environment's GCS bandwidth"
}

cat(sprintf(
    "io_threads=%d : %.1fs (%.2f GB/s)\nio_threads=%d : %.1fs (%.2f GB/s)  %.2fx\n-> %s\n",
    default_io, t_default, gb_total / t_default,
    more_io,    t_more,    gb_total / t_more,    speedup, verdict))
flush.console()
```

**Column projection + predicate pushdown + bounded `collect()`.** `arrow` reads only the selected columns and matching row groups across the shards, in parallel, then materializes a *bounded* result into R:

```r
big_trips <- ds %>%
    select(trip_start_timestamp, trip_miles, fare, trip_total, payment_type) %>%
    filter(trip_miles > 50) %>%      # long trips only -> a small slice of 212M rows
    collect()
dim(big_trips); head(big_trips, 3)
```

## Comparison & Notes

| Pattern | Table | Reads via | Parallel | Scale shown |
|---|---|---|---|---|
| A — `bigrquery` / `DBI` / `dplyr`+`dbplyr` | native | REST + query engine | No (server-side SQL) | small / bounded |
| B — `bigrquerystorage` | native | Storage Read API (Arrow) | Yes (streams) | large bounded slice |
| C — `arrow::open_dataset()` | Iceberg managed | Parquet in your bucket | **Yes — multithreaded across shards** | all 212M rows, scanned in parallel |

**Native vs. Iceberg managed.** A native table is read *through BigQuery* (query engine or Storage Read API). An Iceberg managed table is *also* queryable that way — but because its bytes are open Parquet shards in your bucket, R can read them **directly and in parallel** with `arrow` (pattern C), skipping BigQuery entirely. The many-file layout is what lets `arrow` spread the scan across cores.

**Why not `duckdb` in the middle?** `arrow` already provides a `dplyr` backend with lazy evaluation, projection, predicate pushdown, and multithreaded scans straight over Parquet; `bigrquerystorage` already returns Arrow batches from BigQuery. The extra engine/hop is unnecessary here.

**Memory.** At 212M rows, never `collect()` the whole table — aggregate or filter first (as above). `arrow`'s lazy scan keeps the heavy lifting off R's heap until you ask for a bounded result.
