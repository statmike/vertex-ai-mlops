# R - Dataproc Serverless Spark-R

Runtime: **Dataproc Serverless (Spark)**.  Run the shared workflow distributed across a Spark cluster using [SparkR](https://spark.apache.org/docs/latest/sparkr.html) - submitted as a [Dataproc Serverless](https://cloud.google.com/dataproc-serverless/docs/overview) batch, no cluster to manage.  This is the runtime to reach for when the data outgrows a single machine: the read and the model fit (`spark.glm`) run in parallel across executors.

The Spark version of the workflow is [`code/sparkr.R`](./code/sparkr.R).  It reads BigQuery directly through the Spark BigQuery connector (Storage API / Arrow) - no CSV, no driver-side download.

> Python kernel drives the Dataproc API here.  For the same SparkR task shown alongside PySpark, see [ml-training/dataproc-serverless.ipynb](../../data%2Bai/overview/ml-training/readme.md).

---
Part of the **R** series - see the [series guide](./readme.md). This series runs *one shared workflow* (a logistic-regression fraud-detection GLM on `bigquery-public-data.ml_datasets.ulb_fraud_detection`) across multiple Google Cloud runtimes so you can compare them for your own work.

---
**Prerequisites**
- Run [R - Notebook Workflow](./R%20-%20Notebook%20Workflow.ipynb) first to create the `*_prepped` table (with the `splits` column) that the Spark job reads.

> **Spark, PySpark, Spark-R — what's the relationship?**  [Apache Spark](https://spark.apache.org/) is the distributed compute *engine* (a Scala/JVM core that splits data and work across executors).  It exposes that one engine through several **language APIs** that all compile to the same execution plan: **Scala/Java** (native), **PySpark** (Python), **SparkR** (R — used here, ships with Spark), and **Spark SQL**.  So PySpark and SparkR are *siblings* — different front-ends over the same engine; `spark.glm` here and `pyspark.ml` logistic regression both drive Spark MLlib on the JVM (which is why this run's results match the other runtimes).  On **Dataproc Serverless** the batch *type* selects the API — `spark_r_batch` (this notebook), `pyspark_batch`, `spark_batch` (Scala/Java jar), or `spark_sql_batch`.  (Note: **SparkR** is Apache's R API; [`sparklyr`](https://spark.posit.co/) is a separate dplyr-style R interface to the same engine.)

## Setup

````python
PROJECT_ID = 'statmike-mlops-349915'
PROJECT_ID
````

````python
REGION = 'us-central1'
SERIES = 'r'
EXPERIMENT = 'dataproc-serverless'

# the prepped table created in the Notebook Workflow (note: dataset is the series 'r')
BQ_TABLE_FQN = f'{PROJECT_ID}.r.bigquery-data_prepped'
VAR_TARGET = 'Class'

GCS_BUCKET = PROJECT_ID
````

````python
from google.cloud import storage, dataproc_v1
from datetime import datetime

gcs = storage.Client(project=PROJECT_ID)
batch_client = dataproc_v1.BatchControllerClient(
    client_options={"api_endpoint": f"{REGION}-dataproc.googleapis.com:443"}
)
TIMESTAMP = datetime.now().strftime("%Y%m%d%H%M%S")
````

## The SparkR Script

Distributed version of the shared workflow - `read.df(source="bigquery")`, `spark.glm`, then a confusion matrix.  Review it:

> **Saving the Spark model.**  This script trains and evaluates but does not persist a model — the summary and confusion matrix land in the driver logs.  To save the fitted model, note that `spark.glm` returns a **Spark MLlib model**, *not* a base-R `glm`, so you do **not** use `saveRDS` (and the trim + `plumber` serving path in [R - Serving Predictions](./R%20-%20Serving%20Predictions.ipynb) does not apply directly).  Use SparkR's `write.ml()`, which writes a **directory** of metadata + Parquet (the destination is a path, not a single file — a GCS path works directly):
>
> ```r
> # add near the end of code/sparkr.R, before sparkR.session.stop()
> model_dir <- if (length(args) >= 3) args[3] else "/tmp/spark-glm-model"
> write.ml(model, model_dir, overwrite = TRUE)   # e.g. gs://<bucket>/r/dataproc-serverless/model
> cat("Saved Spark model to", model_dir, "\n")
> ```
>
> Then pass the output path as a third batch argument in the submit cell below (`args=[BQ_TABLE_FQN, VAR_TARGET, "gs://<bucket>/r/dataproc-serverless/model"]`).  Reload it in any SparkR session with `read.ml(model_dir)`.  Two consequences worth knowing: a Spark MLlib model is served *on Spark* (e.g. [Batch Inference With Dataproc](../../MLOps/Serving/Batch/Batch%20Inference%20With%20Dataproc.ipynb)), not by the single-container `plumber` endpoint; and it is **not interchangeable** with the `.rds` the other runtimes produce, even though the fitted coefficients match.

## Network Prerequisite (one-time)

Dataproc Serverless needs the subnet to allow internal traffic and (for Google APIs) Private Google Access.  Enable both on the `default` subnet for this region:

````python
status = !gcloud compute networks subnets describe default --region={REGION} --format="get(privateIpGoogleAccess)"
if status and status[0] == 'False':
    !gcloud compute networks subnets update default --region={REGION} --enable-private-ip-google-access
print('Private Google Access enabled')
````

````python
# allow internal ingress (idempotent)
!gcloud compute firewall-rules create allow-internal-ingress \
    --network=default --direction=ingress --action=allow \
    --source-ranges=10.128.0.0/9 --rules=all 2>/dev/null; echo "ok"
````

## Submit The Spark-R Batch

Stage the script to GCS:

````python
bucket = gcs.bucket(GCS_BUCKET)
blob = bucket.blob(f'{SERIES}/{EXPERIMENT}/{TIMESTAMP}/sparkr.R')
blob.upload_from_filename('./code/sparkr.R')
main_uri = f'gs://{GCS_BUCKET}/{blob.name}'
main_uri
````

Submit the batch, passing the prepped table FQN and target column as args:

````python
operation = batch_client.create_batch(
    parent=f'projects/{PROJECT_ID}/locations/{REGION}',
    batch=dataproc_v1.Batch(
        spark_r_batch=dataproc_v1.SparkRBatch(
            main_r_file_uri=main_uri,
            args=[BQ_TABLE_FQN, VAR_TARGET],
        )
    ),
    batch_id=f'{SERIES}-{EXPERIMENT}-{TIMESTAMP}',
)
result = operation.result()   # waits for the batch to finish
result.state
````

````python
print(f"Review the batch (logs include the model summary + confusion matrix):")
print(f"https://console.cloud.google.com/dataproc/batches/{REGION}/{result.name.split('/')[-1]}/monitoring?project={PROJECT_ID}")
````
