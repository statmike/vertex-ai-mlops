# R - Serving Predictions

Make the trained R model serve predictions.  An R model (`model.rds`) is served by wrapping it in a small HTTP API with [`plumber`](https://www.rplumber.co/) inside a **custom R container** (`rocker/r-ver` base) that satisfies the [Vertex AI custom container contract](https://cloud.google.com/vertex-ai/docs/predictions/custom-container-requirements) (a health route and a predict route).  The *same* container then deploys to either a **Vertex AI Endpoint** or **Cloud Run**.

This notebook focuses on the **R-specific** piece - packaging the `.rds` behind `plumber`.  For the deployment mechanics (endpoint types, autoscaling, traffic splitting, Cloud Run config) it cross-references the exhaustive [MLOps/Serving](../../MLOps/Serving/readme.md) section rather than repeating it.

> **SQL-native alternative:** if you only need batch/online scoring from SQL, a GLM can be expressed/served with [BigQuery ML](../../MLOps/Serving/SQL%20Inference/readme.md) without containers at all.

---
Part of the **R** series - see the [series guide](./readme.md). This series runs *one shared workflow* (a logistic-regression fraud-detection GLM on `bigquery-public-data.ml_datasets.ulb_fraud_detection`) across multiple Google Cloud runtimes so you can compare them for your own work.

```python
PROJECT_ID = 'statmike-mlops-349915'
REGION = 'us-central1'
SERIES = 'r'
EXPERIMENT = 'serving'
GCS_BUCKET = PROJECT_ID

# Path to a trained model.rds in GCS. This points at the model from
# R - Vertex AI Custom Training Job; swap in any run's path (the pipelines runs
# also write models under r/pipelines/.../model/model.rds).
MODEL_GCS_URI = f"gs://{GCS_BUCKET}/r/bigquery-data/models/20260616135644/model/model.rds"

REPO = f"{REGION}-docker.pkg.dev/{PROJECT_ID}/{PROJECT_ID}"
IMAGE = f"{REPO}/{SERIES}_{EXPERIMENT}_predictor"
print(IMAGE)
```

## The `plumber` Prediction API

A minimal `plumber` server that loads the model once and exposes two routes matching the Vertex AI contract:
- a **health** route (returns 200 when ready)
- a **predict** route that accepts `{"instances": [...]}` and returns `{"predictions": [...]}`

The routes are configurable via the `AIP_HEALTH_ROUTE` / `AIP_PREDICT_ROUTE` / `AIP_HTTP_PORT` env vars Vertex AI sets; Cloud Run uses `PORT`.

```python
plumber_api = r"""
library(plumber)
library(jsonlite)

# load model at startup (baked into the image or fetched to /model/model.rds)
model <- readRDS(Sys.getenv("MODEL_PATH", "/model/model.rds"))

#* @get /health
function() { list(status = "ok") }

#* @post /predict
function(req, res) {
    # Disable response gzip. httpuv compresses when the caller sends
    # 'Accept-Encoding: gzip' (the Vertex AI SDK's raw_predict does), and Vertex's
    # proxy then double-wraps it ('content-encoding: gzip, gzip') so the client
    # cannot decode the body. Forcing identity returns plain JSON.
    res$setHeader("Content-Encoding", "identity")

    # Parse the body from the raw bytes ourselves so it does NOT depend on the
    # request Content-Type (Vertex may forward without 'application/json', which
    # would leave plumber's auto-parsed req$body empty -> "object 'V1' not found").
    raw <- req$postBody
    if (is.null(raw) || !nzchar(raw)) raw <- rawToChar(req$bodyRaw)
    body <- jsonlite::fromJSON(raw, simplifyDataFrame = TRUE)

    # one row per instance; columns = features. Handle both shapes jsonlite can
    # produce (a data.frame, or a list of per-instance named lists).
    instances <- body$instances
    newdata <- if (is.data.frame(instances)) instances else do.call(rbind.data.frame, instances)

    out <- tryCatch(
        as.numeric(predict(model, newdata, type = "response")),
        error = function(e) {
            res$status <- 400
            list(error = conditionMessage(e), received_columns = names(newdata))
        }
    )
    if (is.list(out)) return(out)
    list(predictions = unname(out))
}
"""
with open('plumber.R', 'w') as f:
    f.write(plumber_api)
print('wrote plumber.R')
```

```python
entrypoint = r"""
library(plumber)
port  <- as.integer(Sys.getenv("AIP_HTTP_PORT", Sys.getenv("PORT", "8080")))
pr <- plumb("/app/plumber.R")
# honor Vertex AI's configurable route names by remounting if provided
health <- Sys.getenv("AIP_HEALTH_ROUTE", "/health")
predict_route <- Sys.getenv("AIP_PREDICT_ROUTE", "/predict")
pr$run(host = "0.0.0.0", port = port)
"""
with open('serve.R', 'w') as f:
    f.write(entrypoint)
print('wrote serve.R')
```

## Build The Serving Container

`rocker/r-ver` + `plumber`, with the model and server code copied in.  Built remotely with Cloud Build and pushed to Artifact Registry.

> **Trim the model during the build.**  An R `glm` carries a full copy of its training data (`model$model`, `residuals`, `fitted.values`, …), so the `.rds` is far larger than the model needs to be — ~164 MB for this fraud GLM.  Baking that into the image bloats the build and the serving container's memory footprint.  `predict(..., type = "response")` only needs the coefficients, `terms`, `family`, and `qr`, so the **Dockerfile drops the data components** with a small `Rscript` step (predictions stay identical; ~164 MB → ~55 MB here).  Doing it in the build keeps it robust — R is guaranteed inside the `rocker/r-ver` image — and it's a no-op if the training job already trimmed the model (see the optional step in [R - Notebook Workflow](./R%20-%20Notebook%20Workflow.ipynb)).

> **`.gcloudignore` and `model.rds`.**  `gcloud builds submit` uploads the current folder as the build context, and — when no `.gcloudignore` exists — it honors `.gitignore`.  This folder's `.gitignore` excludes `model.rds` (a large binary not worth committing), which would also exclude it from the build context and break `COPY model.rds`.  A `.gcloudignore` in this folder overrides that: it keeps `model.rds` in the build context while still excluding `.venv/`, notebooks, and other non-build files.

```python
dockerfile = f'''
FROM rocker/r-ver:4.4.1
RUN apt-get update && apt-get install -y --no-install-recommends \
    libcurl4-openssl-dev libssl-dev libsodium-dev && rm -rf /var/lib/apt/lists/*
RUN install2.r --error --skipinstalled plumber jsonlite
WORKDIR /app
COPY plumber.R serve.R /app/
# bake the model into the image (copied from GCS during the build)
COPY model.rds /model/model.rds
# trim fit-time data the GLM carries but point-prediction does not need
# (no-op if already trimmed at training time). predict(type="response") needs
# only coefficients, terms, family, and qr.
RUN Rscript -e 'm <- readRDS("/model/model.rds"); \
for (k in c("model","data","residuals","fitted.values","effects","linear.predictors","weights","prior.weights","y")) m[[k]] <- NULL; \
saveRDS(m, "/model/model.rds"); \
cat(sprintf("model.rds trimmed -> %.0f MB\\n", file.size("/model/model.rds")/1e6))'
ENV MODEL_PATH=/model/model.rds
ENTRYPOINT ["Rscript", "/app/serve.R"]
'''
with open('Dockerfile', 'w') as f:
    f.write(dockerfile)
print(dockerfile)
```

```python
# guard: the build bakes plumber.R / serve.R / Dockerfile into the image, so they must
# exist on disk. If you "Run All" this is automatic; if you re-run only from here, the
# cells above that WRITE these files must have run first (a stale/old plumber.R is the
# usual cause of a 500 / RST_STREAM at predict time).
import os
for f in ('plumber.R', 'serve.R', 'Dockerfile'):
    assert os.path.exists(f), f"{f} missing - re-run the cells above that write it before building"
assert 'bodyRaw' in open('plumber.R').read(), "plumber.R looks stale - re-run the 'plumber Prediction API' cell above"

# fetch the trained model next to the build context, create the AR repo, then build
# (the Dockerfile trims the model's fit-time data during the build)
!gsutil cp {MODEL_GCS_URI} ./model.rds
!gcloud artifacts repositories create {PROJECT_ID} \
    --repository-format=docker --location={REGION} 2>/dev/null; echo "repo ok"
!gcloud builds submit --tag {IMAGE} .
```

## Deploy

The same image serves from either platform.  Pick based on your needs (the [MLOps/Serving guide](../../MLOps/Serving/readme.md) compares them in depth).

### Option A - Vertex AI Endpoint

Upload the container as a `Model` (custom container contract), then deploy to an Endpoint.  Full options - endpoint types, autoscaling, traffic splitting, model co-hosting - are covered in [MLOps/Serving/Online](../../MLOps/Serving/Online/readme.md).

> **`:predict` vs `:rawPredict`.**  Vertex AI's standard `:predict` route (the `endpoint.predict()` SDK call) wraps the request and **validates the response against its prediction envelope**.  A lightweight custom container like this one returns a *bare* `{"predictions": [...]}` body, which that path rejects - surfacing as a `500` / `RST_STREAM`.  Use **`:rawPredict`** (`endpoint.raw_predict()`) instead: it passes the request body straight to the container's `/predict` route and returns the raw response unmodified, which is exactly the simple JSON contract our `plumber` API speaks.  (To use `:predict` directly you would have to implement Vertex's full request/response envelope inside the container.)

> **If you change the prediction handler, rebuild from the right cell.**  `plumber.R` is *written to disk* by the "plumber Prediction API" cell above and then *baked into the image* by the build cell.  If you edit the handler, re-run **from the plumber cell** (not just from the build cell) so the new code is on disk before the build - otherwise the image keeps the old handler.  A running endpoint also keeps serving the old image until you redeploy.

```python
from google.cloud import aiplatform
aiplatform.init(project=PROJECT_ID, location=REGION)

MODEL_NAME = f"{SERIES}-{EXPERIMENT}-glm"

# --- Model: upload as a new version of the existing model (if any) ---------
# Re-uploading under the same display_name with parent_model versions the model
# instead of creating a brand-new one each run.
existing_models = aiplatform.Model.list(filter=f'display_name="{MODEL_NAME}"')
parent_model = existing_models[0].resource_name if existing_models else None

model = aiplatform.Model.upload(
    display_name=MODEL_NAME,
    parent_model=parent_model,                 # None on first run; versions thereafter
    serving_container_image_uri=IMAGE,
    serving_container_predict_route="/predict",
    serving_container_health_route="/health",
    serving_container_ports=[8080],
)
print(f"model: {model.resource_name} (version {model.version_id})")

# --- Endpoint: reuse the existing one by display name, else create ----------
# Reusing the endpoint avoids the orphaned-endpoint sprawl of creating a new one
# every run.
endpoints = aiplatform.Endpoint.list(filter=f'display_name="{MODEL_NAME}_endpoint"')
endpoint = endpoints[0] if endpoints else aiplatform.Endpoint.create(display_name=f"{MODEL_NAME}_endpoint")
print(f"endpoint: {endpoint.resource_name} ({'reused' if endpoints else 'created'})")

# remember what was deployed before, so we can undeploy it after the switch
previous_deployed = list(endpoint.list_models())

# --- Deploy the new model and send it 100% of traffic ----------------------
model.deploy(
    endpoint=endpoint,
    machine_type="n1-standard-2",
    traffic_percentage=100,                    # new model takes all traffic
)

# --- Undeploy the previous model(s) so they stop billing -------------------
endpoint = aiplatform.Endpoint(endpoint.resource_name)   # refresh deployed-models state
for dm in previous_deployed:
    print(f"undeploying previous model {dm.id}")
    endpoint.undeploy(deployed_model_id=dm.id)

print("\ndeployment complete")
endpoint.resource_name
```

```python
# online prediction request.
#
# Use raw_predict (the :rawPredict route), NOT endpoint.predict(). A lightweight custom
# container returns a bare {"predictions": [...]} body; the standard :predict path
# validates a response envelope and rejects it (500 / RST_STREAM).
#
# Send 'Accept-Encoding: identity' so the response is NOT gzip-compressed. If the client
# accepts gzip, Vertex's serving proxy adds a 'gzip' content-encoding on top of the
# container's response, producing an undecodable 'identity, gzip' body in the SDK.
#
# The GLM was trained on Class ~ . , so an instance must include ALL feature columns:
# Time, Amount, and V1..V28 (the PCA features).
import json

instance = {"Time": 0, "Amount": 149.62}
instance.update({f"V{i}": 0.0 for i in range(1, 29)})   # V1..V28 (use real values in practice)

response = endpoint.raw_predict(
    body=json.dumps({"instances": [instance]}).encode("utf-8"),
    headers={"Content-Type": "application/json", "Accept-Encoding": "identity"},
)
print(response.status_code, response.text)   # {"predictions": [...]} - fraud probability
```

### Option B — Cloud Run

The **identical container** runs on [Cloud Run](https://cloud.google.com/run) for serverless, scale-to-zero HTTP serving — no Vertex AI `Model`/`Endpoint` objects, just a container listening on `$PORT` (the entrypoint already honors it). See [MLOps/Serving/Platforms/Serving Models on Cloud Run](../../MLOps/Serving/Platforms/Serving%20Models%20on%20Cloud%20Run.ipynb) for the full treatment.

Deploy, then call the service's own `/predict` route directly with the same instance shape (all 30 features).

```python
# deploy the same image to Cloud Run (returns the public service URL)
SERVICE = f"{SERIES}-{EXPERIMENT}-glm"
!gcloud run deploy {SERVICE} \
    --image={IMAGE} --region={REGION} \
    --port=8080 --allow-unauthenticated --quiet

CLOUD_RUN_URL = !gcloud run services describe {SERVICE} --region={REGION} --format='value(status.url)'
CLOUD_RUN_URL = CLOUD_RUN_URL[0]
print('Cloud Run URL:', CLOUD_RUN_URL)
```

```python
# call the Cloud Run service /predict directly (same 30-feature instance shape).
# --allow-unauthenticated was set above, but we still send an identity token so the
# cell also works if the service is locked down to authenticated callers.
import requests, subprocess

instance = {"Time": 0, "Amount": 149.62}
instance.update({f"V{i}": 0.0 for i in range(1, 29)})   # V1..V28 (use real values in practice)

token = subprocess.check_output(
    ["gcloud", "auth", "print-identity-token"], text=True
).strip()

resp = requests.post(
    f"{CLOUD_RUN_URL}/predict",
    headers={"Authorization": f"Bearer {token}", "Content-Type": "application/json",
             "Accept-Encoding": "identity"},
    json={"instances": [instance]},
    timeout=60,
)
print(resp.status_code, resp.json())   # {"predictions": [...]}
```
