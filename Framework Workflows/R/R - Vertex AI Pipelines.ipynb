# R - Vertex AI Pipelines

Runtime: **Vertex AI Pipelines (KFP)**.  Run the shared GLM workflow as a step in an orchestrated, reproducible pipeline — the building block for MLOps DAGs (scheduled, triggered, cached, with artifact lineage).

This notebook doubles as a **mini KFP tutorial** built around *two ways to get R code into a pipeline step*, then a pipeline that runs both together:

| | Approach A — code as a pipeline input | Approach B — code in the image |
|---|---|---|
| **How** | The script text is a **pipeline parameter**, decoded and run inside a stock image | `code/train.R` is **baked into a custom image**; the step just runs it |
| **Image** | plain `rocker/r-ver`; packages installed **at runtime** | `rocker/r-ver` + `bigrquery` prebuilt (our trainer image) |
| **Change the code** | edit a parameter — no rebuild | rebuild the image |
| **Change packages** | runtime install (slower per run) | rebuild the image |
| **Best for** | fast iteration / ad-hoc scripts | stable, fast, production steps |

We run each approach **on its own** (a component *is* a pipeline), then wire both into **one `@dsl.pipeline`** as parallel branches — with the R script passed in as a pipeline parameter.

> Python kernel drives KFP here.  For this R component shown beside Python/BQML/Dataproc steps in a multi-branch pipeline, see [ml-training/kfp-orchestration.ipynb](../../data%2Bai/overview/ml-training/readme.md).

## KFP Basics

[Kubeflow Pipelines (KFP)](https://www.kubeflow.org/docs/components/pipelines/v2/introduction/) describes an ML workflow as a directed graph of **components**. [Vertex AI Pipelines](https://cloud.google.com/vertex-ai/docs/pipelines/introduction) is a fully managed, serverless runner for those pipelines — no Kubernetes cluster to manage.

> For a deeper treatment of pipelines on Vertex AI — components, control flow, scheduling, and patterns — see the repo's [MLOps / Orchestration With Pipelines](../../MLOps/Pipelines/readme.md) section. This notebook focuses on the R-specific angle.

| Concept | What it is |
|---|---|
| **Component** | A self-contained unit of work that runs in its own container. |
| **Pipeline** | A DAG of components wired by inputs/outputs. Compiles to a portable YAML spec. |
| **Artifact** | A typed output (dataset, model, metrics) tracked in Vertex AI ML Metadata for lineage. |
| **Task** | One invocation of a component within a run, with concrete parameter values. |

Why run R this way: **isolation** (the R code runs in its own container, no dependency clash with the Python driver), **reproducibility** (the compiled YAML is a complete spec), **caching** (re-runs skip steps whose inputs are unchanged), and **lineage** (artifacts are tracked end-to-end).

`@dsl.component` wraps a **Python** function as a component; **`@dsl.container_component`** runs **any container, any language** — the right fit for R. Both approaches below use `@dsl.container_component`; they differ only in *where the R code comes from*.

---
## Approach A — Code as a Pipeline Input

Start with the most flexible pattern: the R script is **not** in the image — it travels as a **pipeline parameter**. We base64-encode the local [`code/train.R`](./code/train.R), pass the string in, and the step (on a **stock `rocker/r-ver`** image) decodes it, installs the package it needs at runtime, and runs it. Change the workflow by editing a parameter — no image to rebuild.

**Trade-off:** maximum flexibility, but each run pays for a runtime package install. We install only `bigrquery` (the script treats `bigrquerystorage` as optional and falls back to the REST read path when it is absent), which keeps the install lighter.

```python
# the R code becomes a parameter value — base64 so it survives as a single string arg
R_SCRIPT_B64 = base64.b64encode(open('code/train.R', 'rb').read()).decode()
print('encoded code/train.R ->', len(R_SCRIPT_B64), 'base64 chars')
```

Define the component. We set `AIP_MODEL_DIR` so `train.R` writes `model.rds` through the Cloud Storage FUSE mount (`/gcs/...`):

```python
@dsl.container_component
def r_glm_injected(
    project_id: str, bq_dataset: str, bq_table: str,
    var_target: str, var_omit: str, model_dir: str, r_script_b64: str,
):
    """Approach A: decode the R script from a parameter and run it on a stock image."""
    return dsl.ContainerSpec(
        image=STOCK_IMAGE,
        command=['bash', '-c'],
        args=[dsl.ConcatPlaceholder(items=[
            # system libs for bigrquery + its deps: curl/ssl/xml, plus libuv (for the 'fs' package).
            # bigrquerystorage is optional - the script falls back to the REST read path without it.
            'apt-get update -qq && apt-get install -y -qq libcurl4-openssl-dev libssl-dev libxml2-dev libuv1-dev && ',
            'Rscript -e \'install.packages("bigrquery", repos="https://cloud.r-project.org")\' && ',
            # decode the injected script, then run it
            'echo ', r_script_b64, ' | base64 -d > train.R && ',
            'export AIP_MODEL_DIR=', model_dir, ' && ',
            'Rscript train.R ',
            project_id, ' ', bq_dataset, ' ', bq_table, ' ', var_target, ' ', var_omit,
        ])],
    )

print('Approach A component defined — image:', STOCK_IMAGE)
```

A component *is* a compilable pipeline — no `@dsl.pipeline` wrapper needed. Compile and run it on its own; note `r_script_b64` is an input value:

```python
compiler.Compiler().compile(r_glm_injected, 'r-injected.yaml')
MODEL_DIR_A = f"gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{TIMESTAMP}/injected/model"

job_injected = aiplatform.PipelineJob(
    display_name=f'{SERIES}-injected-{TIMESTAMP}',
    template_path='r-injected.yaml',
    parameter_values=dict(
        project_id=BQ_PROJECT, bq_dataset=BQ_DATASET, bq_table=BQ_TABLE,
        var_target=VAR_TARGET, var_omit=','.join(VAR_OMIT),
        model_dir=MODEL_DIR_A, r_script_b64=R_SCRIPT_B64,
    ),
    pipeline_root=PIPELINE_ROOT, enable_caching=False,
)
job_injected.submit()
print('submitted:', job_injected._dashboard_uri())
```

```python
job_injected.wait()
print('state:', job_injected.state)
print('model.rds exists:', storage.Blob.from_string(f"{MODEL_DIR_A}/model.rds", client=gcs).exists())
```

## Approach B — Code Baked Into the Image

The flip side: bake the code and packages into a **custom image** so the step is fast and self-contained. This runs our shared [`code/train.R`](./code/train.R) from inside the trainer image the [Custom Training Job](./R%20-%20Vertex%20AI%20Custom%20Training%20Job.ipynb) builds — `rocker/r-ver` with `bigrquery` + `bigrquerystorage` baked in (Storage Read API fast path works) and the script copied to `/root/code/train.R`. The component definition is tiny: point at the image, run the script.

**Trade-off:** fast and reproducible, but changing the code (or packages) means rebuilding the image.

Build the trainer image **only if it is missing** (a no-op if you ran the Custom Training notebook):

```python
# create the Artifact Registry repo (idempotent)
!gcloud artifacts repositories create {PROJECT_ID} \
    --repository-format=docker --location={REGION} \
    --description="Docker images for the R series" 2>/dev/null; echo "repo ok"

exists = !gcloud artifacts docker images list {REPO} --include-tags --filter="package:{BAKED_IMAGE}" --format="value(package)" 2>/dev/null
if any(BAKED_IMAGE in line for line in exists):
    print(f'trainer image already exists: {BAKED_IMAGE}')
else:
    print(f'building trainer image: {BAKED_IMAGE}')
    cb = cloudbuild_v1.CloudBuildClient()
    bucket = gcs.bucket(GCS_BUCKET)
    SOURCEPATH = f"{SERIES}/{EXPERIMENT}/build/{TIMESTAMP}"
    bucket.blob(f'{SOURCEPATH}/train.R').upload_from_filename('./code/train.R')
    dockerfile = """
FROM rocker/r-ver:4.4.1
RUN apt-get update && apt-get install -y --no-install-recommends \
    libcurl4-openssl-dev libssl-dev libxml2-dev \
    libprotobuf-dev libgrpc++-dev && rm -rf /var/lib/apt/lists/*
RUN install2.r --error --skipinstalled bigrquery bigrquerystorage
WORKDIR /root
COPY ./*.R ./code/
"""
    bucket.blob(f'{SOURCEPATH}/Dockerfile').upload_from_string(dockerfile)
    build = cloudbuild_v1.Build(steps=[
        {'name': 'gcr.io/cloud-builders/gsutil',
         'args': ['cp', '-r', f'gs://{GCS_BUCKET}/{SOURCEPATH}/*', '/workspace']},
        {'name': 'gcr.io/cloud-builders/docker',
         'args': ['build', '-t', BAKED_IMAGE, '/workspace']},
    ])
    build.images = [BAKED_IMAGE]
    cb.create_build(project_id=PROJECT_ID, build=build).result()
    print('build complete')
```

Define the component — it just runs the baked script (no `r_script_b64` input needed):

```python
@dsl.container_component
def r_glm_baked(
    project_id: str, bq_dataset: str, bq_table: str,
    var_target: str, var_omit: str, model_dir: str,
):
    """Approach B: run code/train.R baked into the trainer image."""
    return dsl.ContainerSpec(
        image=BAKED_IMAGE,
        command=['bash', '-c'],
        args=[dsl.ConcatPlaceholder(items=[
            'export AIP_MODEL_DIR=', model_dir, ' && ',
            'Rscript ./code/train.R ',
            project_id, ' ', bq_dataset, ' ', bq_table, ' ', var_target, ' ', var_omit,
        ])],
    )

print('Approach B component defined — image:', BAKED_IMAGE)
```

Compile and run Approach B on its own:

```python
compiler.Compiler().compile(r_glm_baked, 'r-baked.yaml')
MODEL_DIR_B = f"gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{TIMESTAMP}/baked/model"

job_baked = aiplatform.PipelineJob(
    display_name=f'{SERIES}-baked-{TIMESTAMP}',
    template_path='r-baked.yaml',
    parameter_values=dict(
        project_id=BQ_PROJECT, bq_dataset=BQ_DATASET, bq_table=BQ_TABLE,
        var_target=VAR_TARGET, var_omit=','.join(VAR_OMIT), model_dir=MODEL_DIR_B,
    ),
    pipeline_root=PIPELINE_ROOT, enable_caching=False,
)
job_baked.submit()
print('submitted:', job_baked._dashboard_uri())
```

```python
job_baked.wait()
print('state:', job_baked.state)
print('model.rds exists:', storage.Blob.from_string(f"{MODEL_DIR_B}/model.rds", client=gcs).exists())
```

## A Pipeline That Runs Both

Now wrap both components in one `@dsl.pipeline`. They have no dependency on each other, so KFP runs them as **parallel branches**. The R script is a **pipeline parameter** (`r_script_b64`) flowing into the Approach-A (injected) branch — code as pipeline input, exactly as a real KFP workflow would parameterize it.

```python
@dsl.pipeline(
    name='r-glm-both',
    description='Run the R GLM two ways: code as a pipeline input, and code baked in the image.',
)
def r_glm_both_pipeline(
    project_id: str, bq_dataset: str, bq_table: str,
    var_target: str, var_omit: str,
    model_dir_injected: str, model_dir_baked: str, r_script_b64: str,
):
    r_glm_injected(
        project_id=project_id, bq_dataset=bq_dataset, bq_table=bq_table,
        var_target=var_target, var_omit=var_omit,
        model_dir=model_dir_injected, r_script_b64=r_script_b64,
    ).set_display_name('train-injected')

    r_glm_baked(
        project_id=project_id, bq_dataset=bq_dataset, bq_table=bq_table,
        var_target=var_target, var_omit=var_omit, model_dir=model_dir_baked,
    ).set_display_name('train-baked')

compiler.Compiler().compile(r_glm_both_pipeline, 'r-both.yaml')
print('combined pipeline compiled to r-both.yaml')
```

```python
MODEL_DIR_BOTH_A = f"gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{TIMESTAMP}/both/injected/model"
MODEL_DIR_BOTH_B = f"gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{TIMESTAMP}/both/baked/model"

job_both = aiplatform.PipelineJob(
    display_name=f'{SERIES}-both-{TIMESTAMP}',
    template_path='r-both.yaml',
    parameter_values=dict(
        project_id=BQ_PROJECT, bq_dataset=BQ_DATASET, bq_table=BQ_TABLE,
        var_target=VAR_TARGET, var_omit=','.join(VAR_OMIT),
        model_dir_injected=MODEL_DIR_BOTH_A, model_dir_baked=MODEL_DIR_BOTH_B,
        r_script_b64=R_SCRIPT_B64,
    ),
    pipeline_root=PIPELINE_ROOT, enable_caching=False,
    labels={'series': SERIES, 'experiment': EXPERIMENT},
)
job_both.submit()
print('submitted:', job_both._dashboard_uri())
```

```python
job_both.wait()
```

## Verify

Inspect the per-task status of the combined run, then confirm both branches wrote their model to GCS:

```python
# per-task status from the run itself (robust - no metadata-name lookup needed)
# skip the root DAG task; show just the component tasks
for t in job_both.task_details:
    if t.task_name == job_both.display_name:
        continue
    print(f"{t.task_name:20s} {t.state}")
```

```python
print('combined pipeline state:', job_both.state)
for label, uri in [('injected', MODEL_DIR_BOTH_A), ('baked', MODEL_DIR_BOTH_B)]:
    exists = storage.Blob.from_string(f"{uri}/model.rds", client=gcs).exists()
    print(f'{label:9s} model.rds exists:', exists, '->', f"{uri}/model.rds")
```

## Approach A vs. B — When to Use Which

| | Approach A (code as input) | Approach B (baked image) |
|---|---|---|
| **Code source** | pipeline parameter | inside the image |
| **Image** | stock `rocker/r-ver` | custom (`rocker/r-ver` + packages) |
| **Change code** | edit a parameter | rebuild image |
| **Change packages** | runtime install | rebuild image |
| **Per-run startup** | slower (apt + CRAN install each run) | fast (everything prebuilt) |
| **Best for** | fast iteration, ad-hoc / templated scripts | stable, production, frequently-run steps |

Rule of thumb: **inject the code** while you are still iterating or when a pipeline needs to run user-supplied scripts; **bake the image** once the code and dependencies stabilize.
