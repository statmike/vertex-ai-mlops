# R - Vertex AI Custom Training Job

Runtime: **Vertex AI Custom Training**.  Package the shared workflow script [`code/train.R`](./code/train.R) into a **custom R container** and run it as a managed [CustomJob](https://cloud.google.com/vertex-ai/docs/training/overview) - specify code, container, and compute; pay only for run time; schedule or trigger it as part of MLOps.

We build our **own** container from the community [`rocker/r-ver`](https://rocker-project.org/images/versioned/r-ver.html) base via Cloud Build.  (Vertex AI's prebuilt `r-cpu` deep-learning images still work but are on a deprecation path, so this series does not depend on them.)

> This notebook uses a **Python** kernel for the Vertex AI SDK.  For an all-R driver, the same calls can be made from R with [`reticulate`](https://cran.r-project.org/web/packages/reticulate/vignettes/calling_python.html).  For the same task shown alongside other languages/services, see [ml-training/vertex-training.ipynb](../../data%2Bai/overview/ml-training/readme.md).

## Setup

````python
PROJECT_ID = 'statmike-mlops-349915'
PROJECT_ID
````

````python
REGION = 'us-central1'
SERIES = 'r'
EXPERIMENT = 'bigquery-data'

# BigQuery parameters (the prepped table created in the Notebook Workflow)
BQ_PROJECT = PROJECT_ID
BQ_DATASET = SERIES
BQ_TABLE = f'{EXPERIMENT}_prepped'

# GCS bucket for code staging + model output
GCS_BUCKET = PROJECT_ID

# key columns
VAR_TARGET = 'Class'
VAR_OMIT = ['transaction_id', 'splits']
````

````python
from google.cloud import aiplatform, storage
from google.cloud.devtools import cloudbuild_v1
from datetime import datetime
import os

aiplatform.init(project=PROJECT_ID, location=REGION)
gcs = storage.Client(project=PROJECT_ID)
cb = cloudbuild_v1.CloudBuildClient()

TIMESTAMP = datetime.now().strftime("%Y%m%d%H%M%S")
URI = f"gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}"
REPO = f"{REGION}-docker.pkg.dev/{PROJECT_ID}/{PROJECT_ID}"
IMAGE = f"{REPO}/{SERIES}_{EXPERIMENT}_trainer"
SOURCEPATH = f"{SERIES}/{EXPERIMENT}/build/{TIMESTAMP}"
TIMESTAMP
````

## Build A Custom R Container

We build from `rocker/r-ver` and install the R packages the script needs (`bigrquery`, `bigrquerystorage`), then copy the code in.  Cloud Build runs the build remotely and pushes the image to Artifact Registry.

Create the Artifact Registry Docker repo (idempotent):

````python
!gcloud artifacts repositories create {PROJECT_ID} \
    --repository-format=docker --location={REGION} \
    --description="Docker images for the R series" 2>/dev/null; echo "ok"
````

Stage the script + Dockerfile to GCS for the build:

````python
bucket = gcs.bucket(GCS_BUCKET)
bucket.blob(f'{SOURCEPATH}/train.R').upload_from_filename('./code/train.R')

dockerfile = '''
FROM rocker/r-ver:4.4.1
# libcurl/ssl/xml for bigrquery; libprotobuf + libgrpc++ are RUNTIME deps of the
# bigrquerystorage binary (the BigQuery Storage Read API / Arrow fast path) -
# without them the package fails to load and reads fall back to the slower REST API.
RUN apt-get update && apt-get install -y --no-install-recommends \
    libcurl4-openssl-dev libssl-dev libxml2-dev \
    libprotobuf-dev libgrpc++-dev && rm -rf /var/lib/apt/lists/*
RUN install2.r --error --skipinstalled bigrquery bigrquerystorage
WORKDIR /root
COPY ./*.R ./code/
'''
bucket.blob(f'{SOURCEPATH}/Dockerfile').upload_from_string(dockerfile)
print('staged to', f'gs://{GCS_BUCKET}/{SOURCEPATH}')
````

Run the build (fetch source from GCS → `docker build` → push image):

````python
build = cloudbuild_v1.Build(steps=[
    {'name': 'gcr.io/cloud-builders/gsutil',
     'args': ['cp', '-r', f'gs://{GCS_BUCKET}/{SOURCEPATH}/*', '/workspace']},
    {'name': 'gcr.io/cloud-builders/docker',
     'args': ['build', '-t', IMAGE, '/workspace']},
])
build.images = [IMAGE]

operation = cb.create_build(project_id=PROJECT_ID, build=build)
response = operation.result()   # waits for the build
response.status
````

---
## Submit The Custom Job

The container's entrypoint runs `Rscript ./code/train.R` with the BigQuery parameters as positional args.  Vertex AI sets `AIP_MODEL_DIR` (from `base_output_dir`) and mounts the bucket via Cloud Storage FUSE at `/gcs`, so `train.R` writes `model.rds` straight to that path - no `gcloud`/`gsutil` needed in the container.

````python
CMDARGS = [BQ_PROJECT, BQ_DATASET, BQ_TABLE, VAR_TARGET, ','.join(VAR_OMIT)]

worker_pool_specs = [{
    "replica_count": 1,
    "machine_spec": {"machine_type": "n1-standard-4", "accelerator_count": 0},
    "container_spec": {
        "image_uri": IMAGE,
        "command": ["Rscript", "./code/train.R"],
        "args": CMDARGS,
    },
}]

customJob = aiplatform.CustomJob(
    display_name=f'{SERIES}_{EXPERIMENT}_{TIMESTAMP}',
    worker_pool_specs=worker_pool_specs,
    base_output_dir=f"{URI}/models/{TIMESTAMP}",
    staging_bucket=f"{URI}/models/{TIMESTAMP}",
    labels={'series': SERIES, 'experiment': EXPERIMENT},
)
customJob.run()
````

````python
model_uri = f"{URI}/models/{TIMESTAMP}/model/model.rds"

# verify the training job actually wrote the model
from google.cloud import storage as _storage
_blob = _storage.Blob.from_string(model_uri, client=gcs)
print("model.rds exists:", _blob.exists(), "->", model_uri)

print(f"\nReview the job:\nhttps://console.cloud.google.com/vertex-ai/locations/{REGION}/training/{customJob.resource_name.split('/')[-1]}/cpu?project={PROJECT_ID}")
````
