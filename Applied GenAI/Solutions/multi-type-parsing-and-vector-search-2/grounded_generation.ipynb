# Grounded Generation with Gemini

The retrieval layer (steps 0–10) is mature — 10+ search methods, hybrid RRF, crowding, sparse BM25 — but the system doesn't produce answers. Users must read raw search results themselves.

This notebook adds **grounded generation**: take search results from any method, format them as numbered source context, send to Gemini with a grounding instruction, and produce a cited response. This closes the RAG loop and turns the search pipeline into a full Retrieval-Augmented Generation system.

**What this notebook covers:**
1. Context formatting — numbered sources with chunk_id and source_type
2. System instruction — grounding constraints, citation rules
3. Structured output — Pydantic schema for per-claim citations
4. Generation — Gemini produces structured responses with per-claim source attribution
5. Search method impact — how SemanticSearch, Hybrid RRF, and Crowded search produce different responses
6. Context window management — truncation strategies for large result sets

# Setup

```python
import string
from collections import Counter

import asyncio
import nltk
import pandas as pd
from pydantic import BaseModel
from google.cloud import bigquery, vectorsearch_v1beta
from google import genai
from google.genai.types import GenerateContentConfig
from config import (
    PROJECT_ID, REGION, GEMINI_MODEL, EMBEDDING_MODEL, EMBEDDING_DIMENSIONS,
    BQ_PROJECT, BQ_DATASET, BQ_TABLE_PREFIX,
    VS_COLLECTION_COMBINED,
    BM25_K1, BM25_B, BM25_EPSILON, BM25_NGRAMS,
)

# NLTK resources for BM25 preprocessing
nltk.download('stopwords', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('wordnet', quiet=True)
lemmatizer = nltk.stem.WordNetLemmatizer()
stop_words = set(nltk.corpus.stopwords.words('english'))

# Search clients
search_client = vectorsearch_v1beta.DataObjectSearchServiceClient()
async_search_client = vectorsearch_v1beta.DataObjectSearchServiceAsyncClient()

# BigQuery + Gemini clients
bq = bigquery.Client(project=BQ_PROJECT)
genai_client = genai.Client(vertexai=True, project=PROJECT_ID, location=REGION)

# Collection parent path
COMBINED = f'projects/{PROJECT_ID}/locations/{REGION}/collections/{VS_COLLECTION_COMBINED}'

# Load BM25 vocabulary and model metadata from BigQuery
vocab_table = f'{BQ_PROJECT}.{BQ_DATASET}.{BQ_TABLE_PREFIX}_bm25_vocabulary'
model_table = f'{BQ_PROJECT}.{BQ_DATASET}.{BQ_TABLE_PREFIX}_bm25_model'

vocab_rows = list(bq.query(f'SELECT term, term_index, idf FROM `{vocab_table}`').result())
vocab_map = {row.term: row.term_index for row in vocab_rows}
idf_map = {row.term: row.idf for row in vocab_rows}

model_meta = list(bq.query(f'SELECT * FROM `{model_table}` ORDER BY model_version DESC LIMIT 1').result())[0]

print(f'BM25 model v{model_meta.model_version}: vocab={model_meta.vocab_size}, corpus={model_meta.corpus_size}, avgdl={model_meta.avgdl:.1f}')
print(f'  k1={model_meta.k1}, b={model_meta.b}, epsilon={model_meta.epsilon}, ngrams={model_meta.ngrams}')
print(f'Generation model: {GEMINI_MODEL}')
```

# Search Helpers

Minimal search helpers reused from [search_and_query.ipynb](search_and_query.ipynb) — enough to demonstrate SemanticSearch, Hybrid RRF, and Crowded search as retrieval inputs to generation.

```python
def search_df(results, fields, max_text=150):
    """Format search results as a ranked DataFrame."""
    rows = []
    for r in results:
        data = r.data_object.data
        row = {}
        for f in fields:
            val = data.get(f, '')
            if isinstance(val, str) and len(val) > max_text:
                val = val[:max_text] + '...'
            elif hasattr(val, '__iter__') and not isinstance(val, str):
                val = ', '.join(str(v) for v in val)
            row[f] = val
        rows.append(row)
    df = pd.DataFrame(rows)
    df.index = range(1, len(df) + 1)
    df.index.name = 'rank'
    return df


def hybrid_search(parent, query, weights, top_k=10, fields=None):
    """Run hybrid semantic + text search with RRF reranking."""
    if fields is None:
        fields = ['chunk_id', 'text_content']
    output = vectorsearch_v1beta.OutputFields(data_fields=fields)
    request = vectorsearch_v1beta.BatchSearchDataObjectsRequest(
        parent=parent,
        searches=[
            vectorsearch_v1beta.Search(
                semantic_search=vectorsearch_v1beta.SemanticSearch(
                    search_text=query,
                    search_field='text_content_embedding',
                    task_type='QUESTION_ANSWERING',
                    top_k=top_k,
                    output_fields=output,
                )
            ),
            vectorsearch_v1beta.Search(
                text_search=vectorsearch_v1beta.TextSearch(
                    search_text=query,
                    data_field_names=['text_content'],
                    top_k=top_k,
                    output_fields=output,
                )
            ),
        ],
        combine=vectorsearch_v1beta.BatchSearchDataObjectsRequest.CombineResultsOptions(
            ranker=vectorsearch_v1beta.Ranker(
                rrf=vectorsearch_v1beta.ReciprocalRankFusion(weights=weights)
            )
        ),
    )
    response = search_client.batch_search_data_objects(request)
    if response.results:
        return list(response.results[0].results)
    return []


async def crowded_search(parent, query, crowd_field, crowd_values, top_k_per_value=2, fields=None):
    """Diversity search: top_k results per unique value of a metadata field, run in parallel."""
    if fields is None:
        fields = ['chunk_id', 'source_type', 'text_content']
    output = vectorsearch_v1beta.OutputFields(data_fields=fields)

    async def _search_one(value):
        response = await async_search_client.search_data_objects(
            vectorsearch_v1beta.SearchDataObjectsRequest(
                parent=parent,
                semantic_search=vectorsearch_v1beta.SemanticSearch(
                    search_text=query,
                    search_field='text_content_embedding',
                    task_type='QUESTION_ANSWERING',
                    top_k=top_k_per_value,
                    output_fields=output,
                    filter={crowd_field: {'$eq': value}},
                ),
            )
        )
        return value, list(response.results)

    return dict(await asyncio.gather(*[_search_one(v) for v in crowd_values]))


def chunk_prep(text, ngrams=BM25_NGRAMS):
    """Preprocess text for BM25: lowercase, remove punctuation, tokenize,
    lemmatize, remove stopwords, generate n-grams."""
    text = text.lower().translate(str.maketrans('', '', string.punctuation))
    tokens = nltk.tokenize.word_tokenize(text)
    unigrams = [lemmatizer.lemmatize(t) for t in tokens
                if len(t) > 1 and t not in stop_words]
    if ngrams >= 2:
        all_grams = unigrams.copy()
        for n in range(2, ngrams + 1):
            all_grams.extend(' '.join(g) for g in zip(*[unigrams[i:] for i in range(n)]))
        return all_grams
    return unigrams


def create_bm25_sparse_embedding(text):
    """Compute BM25 sparse embedding for query text using stored vocabulary."""
    tokens = chunk_prep(text)
    term_freqs = Counter(tokens)
    doc_len = len(tokens)
    k1, b, avgdl = model_meta.k1, model_meta.b, model_meta.avgdl

    indices, values = [], []
    for term, freq in term_freqs.items():
        if term not in vocab_map:
            continue
        idf = idf_map[term]
        score = idf * (freq * (k1 + 1)) / (freq + k1 * (1 - b + b * doc_len / avgdl))
        indices.append(vocab_map[term])
        values.append(score)
    return indices, values
```

# Context Formatting

Search results need to be formatted as numbered sources before being sent to Gemini. The format includes chunk_id and source_type metadata so the model can reference specific sources, and so we can trace citations back to their origin.

The function works with results from both `search_data_objects` (flat list) and `batch_search_data_objects` (combined results from RRF) — both produce the same result object structure. For crowded search (dict of grouped results), flatten the groups before calling `format_context`.

An optional `max_chars` parameter truncates individual chunks for context window management (demonstrated in the final section).

```python
def format_context(results, text_field='text_content', max_chars=None):
    """Format search results as numbered source context for grounded generation.

    Args:
        results: List of search result objects (from search_data_objects or batch_search_data_objects).
        text_field: Name of the data field containing the chunk text.
        max_chars: Optional max characters per chunk (truncates with '...' if exceeded).

    Returns:
        Formatted string with numbered sources.
    """
    blocks = []
    for i, r in enumerate(results, 1):
        data = r.data_object.data
        chunk_id = data.get('chunk_id', 'unknown')
        source_type = data.get('source_type', 'unknown')
        text = data.get(text_field, '')
        if max_chars and len(text) > max_chars:
            text = text[:max_chars] + '...'
        blocks.append(f'[Source {i}] (chunk_id: {chunk_id}, type: {source_type})\n{text}')
    return '\n\n'.join(blocks)
```

---
# System Instruction

The system instruction constrains the model to answer only from the provided sources. This is the core grounding mechanism — without it, the model may use its own knowledge rather than the retrieved context.

The default instruction enforces:
- **Source-only answers** — no external knowledge
- **Per-claim citations** — every claim carries its source numbers
- **Synthesis** — combine information across sources rather than summarizing each one
- **Disagreement handling** — present both perspectives as separate claims

Customize the instruction to change response style (e.g., more detailed, more conversational, domain-specific terminology).

```python
DEFAULT_INSTRUCTION = """You are a technical assistant. Answer the user's question based ONLY on
the provided source documents. If the sources don't contain enough
information, say so — do not make up information.

Rules:
- Break your answer into individual claims, each with its supporting source numbers
- Every claim must cite at least one source
- If multiple sources support a claim, include all source numbers
- Synthesize across sources rather than summarizing each one separately
- If sources disagree, present both perspectives as separate claims
- Keep claims concise and technical"""

print(DEFAULT_INSTRUCTION)
```

---
# Response Schema

Using Gemini's [structured output](https://cloud.google.com/vertex-ai/generative-ai/docs/structured-output) with a Pydantic schema to get per-claim citations as structured data. Each `Claim` carries its statement text and the source numbers that support it. This eliminates regex-based citation parsing and gives direct access to which sources support each claim.

The schema is passed to `GenerateContentConfig(response_schema=...)` which constrains Gemini to return JSON matching this structure.

```python
class Claim(BaseModel):
    """A single claim with its supporting source numbers."""
    statement: str
    source_numbers: list[int]


class GroundedResponse(BaseModel):
    """Structured response with per-claim citations."""
    claims: list[Claim]
```

---
# Generate Function

The generation function takes a query and pre-formatted context string, sends them to Gemini with the system instruction and response schema, and returns a `GroundedResponse` with structured per-claim citations. The prompt structure places sources first, then the question — this gives the model full context before it sees what to answer.

```python
def generate_grounded(query, context, system_instruction=DEFAULT_INSTRUCTION, model=None):
    """Generate a grounded response using Gemini with retrieved context.

    Args:
        query: The user's question.
        context: Formatted source context string (from format_context).
        system_instruction: System instruction for grounding constraints.
        model: Gemini model name (defaults to GEMINI_MODEL from config).

    Returns:
        GroundedResponse with per-claim citations.
    """
    response = genai_client.models.generate_content(
        model=model or GEMINI_MODEL,
        contents=f'Sources:\n\n{context}\n\nQuestion: {query}',
        config=GenerateContentConfig(
            system_instruction=system_instruction,
            response_mime_type='application/json',
            response_schema=GroundedResponse,
        ),
    )
    return GroundedResponse.model_validate_json(response.text)
```

---
# Response Helpers

Helpers for working with the structured `GroundedResponse`. `render_response` produces readable text with inline citations from the structured claims. `extract_citations` collects all unique source numbers. `citation_summary` cross-references cited sources against the search results to show which chunks were used and which were available but not cited.

```python
def render_response(grounded):
    """Render structured response as readable text with inline citations."""
    lines = []
    for claim in grounded.claims:
        citations = ''.join(f'[Source {n}]' for n in claim.source_numbers)
        lines.append(f'{claim.statement} {citations}')
    return '\n\n'.join(lines)


def extract_citations(grounded):
    """Extract unique source numbers from structured response."""
    return sorted({n for claim in grounded.claims for n in claim.source_numbers})


def citation_summary(grounded, results, text_field='text_content'):
    """Show which sources were cited vs available.

    Args:
        grounded: GroundedResponse object.
        results: List of search result objects used as context.
        text_field: Name of the data field containing chunk text.

    Returns:
        DataFrame with source number, cited status, chunk_id, source_type, and text preview.
    """
    cited = set(extract_citations(grounded))
    rows = []
    for i, r in enumerate(results, 1):
        data = r.data_object.data
        text = data.get(text_field, '')
        rows.append({
            'source': i,
            'cited': 'yes' if i in cited else '',
            'chunk_id': data.get('chunk_id', ''),
            'source_type': data.get('source_type', ''),
            'text_preview': text[:80] + '...' if len(text) > 80 else text,
        })
    return pd.DataFrame(rows)


def claims_df(grounded):
    """Format structured response as a DataFrame of claims."""
    rows = []
    for claim in grounded.claims:
        rows.append({
            'statement': claim.statement,
            'sources': claim.source_numbers,
        })
    df = pd.DataFrame(rows)
    df.index = range(1, len(df) + 1)
    df.index.name = 'claim'
    return df
```

# Comparison

Same query, three retrieval methods. Compare the source types in context, source types cited, number of claims, and a preview of each response.

```python
def source_types(results):
    """Count source types in a result list."""
    types = [r.data_object.data.get('source_type', 'unknown') for r in results]
    return dict(Counter(types))


def cited_source_types(grounded, results):
    """Count source types of cited sources only."""
    cited_nums = set(extract_citations(grounded))
    types = []
    for i, r in enumerate(results, 1):
        if i in cited_nums:
            types.append(r.data_object.data.get('source_type', 'unknown'))
    return dict(Counter(types))


comparison = pd.DataFrame([
    {
        'method': 'SemanticSearch',
        'sources_in_context': source_types(semantic_results),
        'sources_cited': cited_source_types(response, semantic_results),
        'num_claims': len(response.claims),
        'citations_used': len(extract_citations(response)),
        'response_preview': render_response(response)[:150] + '...',
    },
    {
        'method': 'Hybrid RRF',
        'sources_in_context': source_types(hybrid_results),
        'sources_cited': cited_source_types(response_hybrid, hybrid_results),
        'num_claims': len(response_hybrid.claims),
        'citations_used': len(extract_citations(response_hybrid)),
        'response_preview': render_response(response_hybrid)[:150] + '...',
    },
    {
        'method': 'Crowded',
        'sources_in_context': source_types(crowded_flat),
        'sources_cited': cited_source_types(response_crowded, crowded_flat),
        'num_claims': len(response_crowded.claims),
        'citations_used': len(extract_citations(response_crowded)),
        'response_preview': render_response(response_crowded)[:150] + '...',
    },
])

comparison
```

# Context Window Management

As `top_k` increases or chunks get longer, the combined context can approach Gemini's context window limit. Three strategies to manage this:

1. **Reduce `top_k`** — fewer chunks, less context
2. **Truncate chunks** — use `max_chars` in `format_context` to limit individual chunk length
3. **Summarize before generation** — pre-summarize long chunks (adds latency and cost)

The simplest approach is to combine `top_k` and `max_chars`. Below we demonstrate with `top_k=10` and `max_chars=500` to show the effect on context size and response quality.

```python
# Full context (no truncation)
full_results = hybrid_search(COMBINED, query, weights=[1.0, 1.0], top_k=10, fields=fields)
full_context = format_context(full_results)
approx_tokens_full = len(full_context.split())

# Truncated context
truncated_context = format_context(full_results, max_chars=500)
approx_tokens_truncated = len(truncated_context.split())

print(f'top_k=10, no truncation:     ~{approx_tokens_full:,} words')
print(f'top_k=10, max_chars=500:     ~{approx_tokens_truncated:,} words')
print(f'Reduction:                   {(1 - approx_tokens_truncated / approx_tokens_full) * 100:.0f}%')
```

# Observations

**Structured output:**
- Per-claim citations via Pydantic schema eliminate regex parsing and provide direct access to which sources support each claim
- Each claim is independently traceable — downstream evaluation can check whether a claim is actually supported by its cited source(s)
- The `claims_df` view makes it easy to inspect the response structure and spot uncited or weakly-cited claims
- `render_response` reconstructs readable text with inline `[Source N]` notation from the structured data

**Search method impact on generation:**
- **SemanticSearch** context is dominated by conversational content (Reddit), producing opinion-heavy claims with community-sourced recommendations
- **Hybrid RRF** brings in PDF documentation alongside Reddit discussion, producing claims that include both practical experience and technical reference details
- **Crowded search** forces representation from all source types (reddit, zoom, pdf), producing the most balanced multi-perspective claims

**Citation behavior:**
- The model generally cites the most relevant sources and may skip less relevant ones, even when they're available in context
- Multi-source citations on a single claim appear when multiple sources independently support the same point
- The number of claims varies by response — more diverse context (crowded search) tends to produce more claims

**Context window management:**
- `max_chars` truncation reduces context size while preserving the beginning of each chunk (where the most important structural context usually appears)
- At 10+ sources, truncation becomes important for staying within token limits and keeping generation focused
- The model can still produce good responses from truncated context because the `[Source N]` headers and leading text provide enough signal

**For production use:**
- Crowded or hybrid RRF provides the most reliable grounding — single-method retrieval can produce biased responses
- The system instruction is critical: without it, the model freely mixes retrieved context with its own knowledge
- Structured output enables programmatic quality checks — claims with empty `source_numbers`, source numbers outside the valid range, or zero total claims all indicate grounding failures
