# Inspect Document AI Layout Parser v1.5 Response

Before building the PDF parsing pipeline (step 3c), this notebook processes a single PDF with the [Layout Parser](https://cloud.google.com/document-ai/docs/layout-parse-chunk) and inspects the full response structure. This helps us understand the available fields and design the right BigQuery schema for our chunks.

The Layout Parser v1.5 (`pretrained-layout-parser-v1.5-pro-2025-08-25`) produces two main response objects:
- **`chunked_document`** — chunks with `chunk_id`, `content`, `page_span`, headers/footers, and `source_block_ids`
- **`document_layout`** — hierarchical block structure with `text_block`, `table_block`, and `list_block` types

---
## Setup

```python
import json
from google.cloud import documentai, storage
from config import (
    PROJECT_ID, GCS_BUCKET, GCS_ROOT,
    DOCAI_LOCATION, DOCAI_PROCESSOR_DISPLAY_NAME, DOCAI_PROCESSOR_TYPE, DOCAI_PROCESSOR_VERSION,
)

# Document AI client (uses "us" or "eu" location, not full region)
gcs = storage.Client(project=PROJECT_ID)
docai_client = documentai.DocumentProcessorServiceClient(
    client_options=dict(api_endpoint=f"{DOCAI_LOCATION}-documentai.googleapis.com")
)

print(f'Project: {PROJECT_ID}')
print(f'DocAI location: {DOCAI_LOCATION}')
```

---
## Get or Create Processor

The Layout Parser processor is created once and reused. The processor version `pretrained-layout-parser-v1.5-pro-2025-08-25` is the latest LLM-powered version.

```python
parent = f'projects/{PROJECT_ID}/locations/{DOCAI_LOCATION}'

processor = None
for p in docai_client.list_processors(parent=parent):
    if p.display_name == DOCAI_PROCESSOR_DISPLAY_NAME:
        processor = p
        print(f'Found existing processor: {processor.name}')
        break

if processor is None:
    processor = docai_client.create_processor(
        parent=parent,
        processor=dict(
            display_name=DOCAI_PROCESSOR_DISPLAY_NAME,
            type_=DOCAI_PROCESSOR_TYPE,
            default_processor_version=DOCAI_PROCESSOR_VERSION,
        ),
    )
    print(f'Created processor: {processor.name}')
```

---
## Process a Single PDF

Using online processing with `gcs_document` to quickly process one PDF for inspection. The v1.5 processor supports up to 30 pages for online processing (up from 15 in v1.0). The `create_time_series` PDF has 37 pages, so we use `from_start=30` to stay within the limit. The batch pipeline (`parse_pdf.py`) uses offline batch processing which supports up to 500 pages.

```python
# Use the create_time_series PDF — it's the largest (826KB, 37 pages) and contains images
test_uri = f'gs://{GCS_BUCKET}/{GCS_ROOT}/generated_pdf/bigquery_docs_reference_standard_sql_bigqueryml_syntax_create_time_series.pdf'
print(f'Processing: {test_uri}')

response = docai_client.process_document(
    request=documentai.ProcessRequest(
        name=processor.name,
        gcs_document=documentai.GcsDocument(
            gcs_uri=test_uri,
            mime_type='application/pdf',
        ),
        process_options=documentai.ProcessOptions(
            layout_config=documentai.ProcessOptions.LayoutConfig(
                chunking_config=documentai.ProcessOptions.LayoutConfig.ChunkingConfig(
                    chunk_size=500,
                    include_ancestor_headings=True,
                )
            ),
            # Process first 30 pages (v1.5 online limit is 30 pages)
            from_start=30,
        ),
    )
)

print(f'Response received')
```

---
## Response Overview

The response `document` contains two key objects:
- `chunked_document` — text chunks with semantic boundaries
- `document_layout` — hierarchical block structure

```python
doc = response.document

print(f'chunked_document chunks: {len(doc.chunked_document.chunks)}')
print(f'document_layout blocks:  {len(doc.document_layout.blocks)}')
```

---
## Inspect Chunks

Each chunk represents a semantically coherent section of the document. The Layout Parser respects heading boundaries, tables, and lists when creating chunks.

### Chunk Keys

Convert to dict and discover all available fields across all chunks.

```python
# Convert to dict for inspection (use_integers_for_enums=False for readable enum values)
doc_dict = documentai.Document.to_dict(doc, use_integers_for_enums=False)
chunks = doc_dict['chunked_document']['chunks']

# Collect all keys across all chunks
all_keys = set()
for chunk in chunks:
    all_keys.update(chunk.keys())

print(f'{len(chunks)} chunks with keys: {sorted(all_keys)}')
```

### First Chunk (Full JSON)

Inspect the complete structure of the first chunk.

```python
print(json.dumps(chunks[0], indent=2, default=str))
```

### All Chunks Summary

Show `chunk_id`, page span, content preview, and which optional fields are present for each chunk.

```python
for chunk in chunks:
    cid = chunk.get('chunk_id', '?')
    ps = chunk.get('page_span', {})
    pages = f"p{ps.get('page_start','?')}-{ps.get('page_end','?')}"
    content = chunk.get('content', '')[:100].replace('\n', ' ')
    
    # Flag which optional fields are present
    flags = []
    if chunk.get('page_headers'): flags.append('headers')
    if chunk.get('page_footers'): flags.append('footers')
    if chunk.get('source_block_ids'): flags.append(f"blocks:{len(chunk['source_block_ids'])}")
    
    # Check for any keys we haven't accounted for
    known = {'chunk_id', 'content', 'page_span', 'page_headers', 'page_footers', 'source_block_ids'}
    extra = set(chunk.keys()) - known
    if extra: flags.append(f'NEW:{extra}')
    
    flag_str = f"  [{', '.join(flags)}]" if flags else ''
    print(f'{cid:5s} {pages:8s}{flag_str}  {content}...')
```

### Chunk With Most Fields

Find and display the chunk that has the most non-empty optional fields (headers, footers, source_block_ids, etc.) to see the richest example.

```python
richest = max(chunks, key=lambda c: sum(1 for k, v in c.items() if v and k not in ('chunk_id', 'content')))
print(json.dumps(richest, indent=2, default=str))
```

### Chunk Content Examples

Print the full `content` of a few chunks to see how headings, tables, and lists are represented in markdown.

```python
for chunk in chunks[:3]:
    print(f"--- {chunk['chunk_id']} ---")
    print(chunk['content'])
    print()
```

## Inspect Document Layout Blocks

The `document_layout` provides a hierarchical view of the document structure. Blocks can be `text_block` (paragraphs, headings), `table_block` (tabular data), or `list_block` (ordered/unordered lists). Blocks nest recursively under headings.

```python
blocks = doc_dict['document_layout']['blocks']

# Recursively count block types
def count_block_types(blocks):
    counts = {'text_block': 0, 'table_block': 0, 'list_block': 0}
    type_details = []  # (type, subtype, text_preview)
    for block in blocks:
        for bt in counts:
            if bt in block and block[bt]:
                counts[bt] += 1
                b = block[bt]
                subtype = b.get('type_', '') if bt == 'text_block' else (b.get('type_', '') if bt == 'list_block' else 'table')
                text = b.get('text', '')[:60] if bt == 'text_block' else ''
                type_details.append((bt, subtype, text))
                nested = b.get('blocks', []) or b.get('list_entries', [])
                if nested:
                    sub_counts, sub_details = count_block_types(nested if isinstance(nested, list) else [])
                    for k in counts:
                        counts[k] += sub_counts[k]
                    type_details.extend(sub_details)
    return counts, type_details

counts, details = count_block_types(blocks)
print(f'Top-level blocks: {len(blocks)}')
print(f'All blocks (recursive): {counts}')

# Show text_block subtypes
subtypes = {}
for bt, st, _ in details:
    if bt == 'text_block':
        subtypes[st] = subtypes.get(st, 0) + 1
print(f'Text block subtypes: {subtypes}')
```

### First Block Structure

Show the first top-level block with its nested structure (truncated for readability).

```python
def summarize_block(block, depth=0):
    """Print a readable summary of a block's structure."""
    indent = '  ' * depth
    page = block.get('page_span', {})
    pages = f"p{page.get('page_start','?')}-{page.get('page_end','?')}" if page else ''
    bid = block.get('block_id', '')
    
    if 'text_block' in block and block['text_block']:
        tb = block['text_block']
        text = tb.get('text', '')[:80].replace('\n', ' ')
        btype = tb.get('type_', 'unknown')
        nested = tb.get('blocks', [])
        print(f'{indent}[{bid}] text_block ({btype}) {pages}: "{text}"')
        if nested:
            print(f'{indent}  └─ {len(nested)} nested blocks:')
            for nb in nested[:5]:
                summarize_block(nb, depth + 2)
            if len(nested) > 5:
                print(f'{indent}      ... +{len(nested)-5} more')
    elif 'table_block' in block and block['table_block']:
        tb = block['table_block']
        h = len(tb.get('header_rows', []))
        b = len(tb.get('body_rows', []))
        caption = tb.get('caption', '')
        print(f'{indent}[{bid}] table_block {pages}: {h} header rows, {b} body rows' + (f', caption: "{caption}"' if caption else ''))
    elif 'list_block' in block and block['list_block']:
        lb = block['list_block']
        entries = lb.get('list_entries', [])
        ltype = lb.get('type_', 'unknown')
        print(f'{indent}[{bid}] list_block ({ltype}) {pages}: {len(entries)} entries')

# Show all top-level blocks
for block in blocks:
    summarize_block(block)
```

### Table Block Detail

Find and inspect any table blocks to see how tabular data is structured.

```python
def find_tables(blocks, found=None):
    """Recursively find all table blocks."""
    if found is None: found = []
    for block in blocks:
        if 'table_block' in block and block['table_block']:
            found.append(block)
        for bt in ['text_block', 'list_block']:
            if bt in block and block[bt]:
                nested = block[bt].get('blocks', [])
                if nested:
                    find_tables(nested, found)
    return found

tables = find_tables(blocks)
print(f'{len(tables)} table blocks found')

if tables:
    print('\nFirst table block (full JSON):')
    print(json.dumps(tables[0], indent=2, default=str))
```

### How Tables Appear in Chunks

Find chunks that contain markdown table syntax (`|`) to see how the chunker represents tabular data.

```python
table_chunks = [c for c in chunks if '|' in c.get('content', '') and '|-' in c.get('content', '')]
print(f'{len(table_chunks)} chunks contain table markdown\n')

if table_chunks:
    print(f'--- {table_chunks[0]["chunk_id"]} ---')
    print(table_chunks[0]['content'])
```

### All Block Keys

Discover any new or unexpected fields in the block structure.

```python
def collect_block_keys(blocks, keys=None):
    """Recursively collect all unique keys from blocks."""
    if keys is None: keys = {'top': set(), 'text_block': set(), 'table_block': set(), 'list_block': set()}
    for block in blocks:
        keys['top'].update(block.keys())
        for bt in ['text_block', 'table_block', 'list_block']:
            if bt in block and block[bt]:
                keys[bt].update(block[bt].keys())
                nested = block[bt].get('blocks', [])
                if nested:
                    collect_block_keys(nested, keys)
    return keys

all_block_keys = collect_block_keys(blocks)
for level, k in all_block_keys.items():
    print(f'{level}: {sorted(k)}')
```

### Source Block IDs in Chunks

Each chunk has `source_block_ids` linking back to the layout blocks it was derived from. Inspect this mapping.

```python
chunks_with_sources = [c for c in chunks if c.get('source_block_ids')]
print(f'{len(chunks_with_sources)} of {len(chunks)} chunks have source_block_ids\n')

for chunk in chunks_with_sources[:5]:
    print(f"{chunk['chunk_id']}: source_block_ids = {chunk['source_block_ids']}")
```
