# Embeddings API

Embeddings turn content -- text, images, video, audio -- into vectors so that *meaning* becomes geometry: similar things land close together and can be searched, clustered, classified, and matched. This notebook is the single applied reference for generating embeddings on Google Cloud, spanning three generations of models in one place.

**Model lineage covered here:**

| Generation | Model | Modalities | Default dims | How it is called here |
|---|---|---|---|---|
| Current | `gemini-embedding-2` | text, image, video, audio, documents, **interleaved** | 3072 (Matryoshka 128-3072) | `google-genai` SDK + REST |
| Text (prior) | `text-embedding-005`, `gemini-embedding-001` | text | up to 3072 | `google-genai` SDK + REST |
| Multimodal (prior) | `multimodalembedding@001` | text, image, video (visual only) | 1408 (128/256/512/1408) | REST |

**Forward path:** the `google-genai` SDK (`from google import genai`) is the supported, forward-looking client for Google's generative models. The older `vertexai.*` GenAI modules are deprecating; this notebook uses `google-genai` for SDK calls and raw `requests` to show the underlying REST surface.

**When to use which:** reach for `gemini-embedding-2` for new work -- one model, one shared vector space across modalities, and flexible output dimensions. The prior text and multimodal models are shown for continuity and for workloads already built on them (and for batch, where they are currently the supported path).

> **Model id & region:** `gemini-embedding-2` availability is region-specific. The GA id `gemini-embedding-2` is served from location `us`; the preview id `gemini-embedding-2-preview` is served from regional locations like `us-central1`. This notebook uses `us-central1` + the preview id so a single client reaches the current model *and* every prior model. The Setup cell explains how to switch to the GA id. Confirm what your project has with `gcloud ai models list --region=<LOCATION>`.

**References:**
- Companion math notebook: [`The Math of Similarity.ipynb`](./The%20Math%20of%20Similarity.ipynb)
- Visualization techniques: [`Bring Embeddings to Life With Visualization Techniques.ipynb`](./Bring%20Embeddings%20to%20Life%20With%20Visualization%20Techniques.ipynb)

> **Cost note:** embedding calls are billed per input. The multimodal, matching, and batch sections make many calls -- run them on a subset first. The video/audio and batch cells are guarded so a clean *Restart & Run All* does not incur long-running jobs unless you opt in.

```python
import os, json, base64, io, time
import numpy as np
import requests
import PIL.Image
import matplotlib.pyplot as plt

from google import genai
from google.genai import types
from google.cloud import storage, bigquery
import google.auth
import google.auth.transport.requests

# google-genai client pointed at the Vertex AI backend
client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)

# Supporting clients (batch section)
gcs = storage.Client(project=PROJECT_ID)
bq = bigquery.Client(project=PROJECT_ID)

# Bearer token for the raw REST calls -- reused by every cell in the REST section
def access_token():
    creds, _ = google.auth.default(scopes=['https://www.googleapis.com/auth/cloud-platform'])
    creds.refresh(google.auth.transport.requests.Request())
    return creds.token

# L2-normalize a vector before dot-product (cosine) similarity. gemini-embedding-2 returns
# unit-length vectors at every output dimension, so this is a safe no-op here -- but it keeps
# the matching code correct for any model/dimension that does not normalize.
def l2_normalize(vec):
    vec = np.asarray(vec, dtype=float)
    norm = np.linalg.norm(vec)
    return vec if norm == 0 else vec / norm

print(f'google-genai client ready for {PROJECT_ID} @ {LOCATION}')
```

### Output dimensions (Matryoshka) and normalization

`gemini-embedding-2` uses Matryoshka Representation Learning: request a smaller `output_dimensionality` (recommended 768, 1536, or 3072) to trade a little accuracy for lower storage and faster search. Smaller dimensions are truncations of the full vector.

The cell below sweeps a few sizes and checks each vector's L2 norm. This model returns unit-length vectors at every dimension, so the dot product equals cosine similarity directly -- but verifying the norm is a good habit, since not every model normalizes, and `l2_normalize` (defined in Setup) makes the matching code robust either way.


```python
for dim in [3072, 1536, 768, 256]:
    r = client.models.embed_content(
        model=GEMINI_EMBED,
        contents='a smaller vector still captures the meaning',
        config=types.EmbedContentConfig(output_dimensionality=dim),
    )
    v = np.asarray(r.embeddings[0].values)
    print(f'dim={dim:5d}   L2 norm={np.linalg.norm(v):.4f}')
```

### Task types: symmetric and asymmetric

`task_type` tells the embedding model how a piece of text will be used, which changes how its vector is generated. Choosing the right one is one of the highest-leverage knobs for retrieval quality. `gemini-embedding-2` accepts `task_type` via `EmbedContentConfig`, and so do the prior text models.

**Asymmetric** task types pair *different* types on each side -- what you index differs in intent from what you search with. This is the pattern for retrieval:

| Use case | Document (indexed) | Query (search) |
|---|---|---|
| Search | `RETRIEVAL_DOCUMENT` | `RETRIEVAL_QUERY` |
| Question answering | `RETRIEVAL_DOCUMENT` | `QUESTION_ANSWERING` |
| Fact checking | `RETRIEVAL_DOCUMENT` | `FACT_VERIFICATION` |
| Code retrieval | `RETRIEVAL_DOCUMENT` | `CODE_RETRIEVAL_QUERY` |

**Symmetric** task types use the *same* type on both sides, because both inputs play the same role:

- `SEMANTIC_SIMILARITY` -- how similar two texts are
- `CLASSIFICATION` -- grouping texts by category
- `CLUSTERING` -- organizing texts into clusters

> **Rule of thumb:** when in doubt for retrieval, embed with `RETRIEVAL_DOCUMENT` at index time and `RETRIEVAL_QUERY` at search time.

For the distance measures these vectors are scored with, see [`The Math of Similarity.ipynb`](./The%20Math%20of%20Similarity.ipynb); for the same task-type choices applied inside BigQuery, see [`ml_generate_embedding.ipynb`](../../data%2Bai/bq-ai-functions/functions/ml_generate_embedding/ml_generate_embedding.ipynb).


```python
# Asymmetric retrieval: the query and the document are embedded with DIFFERENT task types.
query = 'What is a balk?'
document = 'A balk is an illegal act by the pitcher with one or more runners on base, entitling all runners to advance one base.'

q = l2_normalize(client.models.embed_content(
    model=GEMINI_EMBED, contents=query,
    config=types.EmbedContentConfig(task_type='RETRIEVAL_QUERY'),
).embeddings[0].values)

d = l2_normalize(client.models.embed_content(
    model=GEMINI_EMBED, contents=document,
    config=types.EmbedContentConfig(task_type='RETRIEVAL_DOCUMENT'),
).embeddings[0].values)

print(f'cosine(query, document) = {float(np.dot(q, d)):.4f}')
```

```python
# The full task-type roster, grouped by whether query and document sides differ.
task_types = {
    'asymmetric (query/document differ)': [
        'RETRIEVAL_QUERY', 'RETRIEVAL_DOCUMENT', 'QUESTION_ANSWERING',
        'FACT_VERIFICATION', 'CODE_RETRIEVAL_QUERY',
    ],
    'symmetric (same on both sides)': [
        'SEMANTIC_SIMILARITY', 'CLASSIFICATION', 'CLUSTERING',
    ],
}
for group, types_list in task_types.items():
    print(group)
    for tt in types_list:
        r = client.models.embed_content(
            model=GEMINI_EMBED, contents='the quick brown fox',
            config=types.EmbedContentConfig(task_type=tt),
        )
        print(f'  {tt:22s} -> {len(r.embeddings[0].values)} dims')
```

## REST from Python (`requests`)

Under the SDK, embeddings are HTTP calls to the Vertex AI API. Calling them directly with `requests` is useful from environments without the SDK and makes the request/response shape explicit. Every call authorizes with a bearer token from `access_token()` (defined in Setup).


### Prior text model -- `:predict`

The publisher `:predict` endpoint takes `instances` (each with `content` and an optional `task_type`) and `parameters` (`outputDimensionality`, `autoTruncate`).


```python
endpoint = (f'https://{LOCATION}-aiplatform.googleapis.com/v1/projects/{PROJECT_ID}'
            f'/locations/{LOCATION}/publishers/google/models/{TEXT_EMBED}:predict')
body = {
    'instances': [{'task_type': 'RETRIEVAL_DOCUMENT', 'content': 'an embedding requested over REST'}],
    'parameters': {'outputDimensionality': 768, 'autoTruncate': True},
}
resp = requests.post(
    endpoint,
    headers={'Authorization': f'Bearer {access_token()}', 'Content-Type': 'application/json'},
    json=body,
)
resp.raise_for_status()
pred = resp.json()['predictions'][0]
print(f"{TEXT_EMBED}: {len(pred['embeddings']['values'])} dims")
```

### Prior multimodal model -- `:predict`

`multimodalembedding@001` accepts `text` and/or `image` (as a GCS `gcsUri` or base64 `bytesBase64Encoded`) per instance, with `dimension` in {128, 256, 512, 1408}. The response returns `textEmbedding` and `imageEmbedding` in the same space.


```python
b64_image = base64.b64encode(image_bytes).decode('utf-8')
endpoint = (f'https://{LOCATION}-aiplatform.googleapis.com/v1/projects/{PROJECT_ID}'
            f'/locations/{LOCATION}/publishers/google/models/{MM_EMBED}:predict')
body = {
    'instances': [{'text': 'a photograph of a dog', 'image': {'bytesBase64Encoded': b64_image}}],
    'parameters': {'dimension': 1408},
}
resp = requests.post(
    endpoint,
    headers={'Authorization': f'Bearer {access_token()}', 'Content-Type': 'application/json'},
    json=body,
)
resp.raise_for_status()
pred = resp.json()['predictions'][0]
print(f"text: {len(pred['textEmbedding'])} dims, image: {len(pred['imageEmbedding'])} dims")
```

### Current model -- `gemini-embedding-2` over REST

`gemini-embedding-2` uses a **different endpoint** from the prior models: `:embedContent` rather than `:predict`. The request body is also different -- `content.parts[]` and `outputDimensionality` sit at the top level (no `instances`/`parameters` wrapper), matching the SDK's `embed_content` call -- and the response is `embedding.values` instead of `predictions[]`.

> The SDK (used earlier) is the recommended path and tracks request-shape changes automatically. This cell shows the underlying HTTP call for when the SDK is not available.


```python
endpoint = (f'https://{LOCATION}-aiplatform.googleapis.com/v1/projects/{PROJECT_ID}'
            f'/locations/{LOCATION}/publishers/google/models/{GEMINI_EMBED}:embedContent')
body = {
    'content': {'parts': [{'text': 'a gemini-embedding-2 vector requested over REST'}]},
    'outputDimensionality': 768,
}
resp = requests.post(
    endpoint,
    headers={'Authorization': f'Bearer {access_token()}', 'Content-Type': 'application/json'},
    json=body,
)
resp.raise_for_status()
values = resp.json()['embedding']['values']
print(f"{GEMINI_EMBED}: {len(values)} dims")
```

## Matching (cross-modal retrieval)

Because every modality shares one space, a **text** query can retrieve **images** and vice-versa. Below, each breed is embedded twice with `gemini-embedding-2` -- once from its text and once from its image -- into two aligned NumPy matrices. Similarity is a dot product on L2-normalized vectors (i.e. cosine).


```python
MATCH_DIM = 768  # smaller vectors keep the local DB light

vector_lookup, text_vecs, image_vecs = [], [], []
for g, group in enumerate(breed_data):
    for b, breed in enumerate(group['breeds']):
        label = f"{group['name']} - {breed['name']}"
        tvec = client.models.embed_content(
            model=GEMINI_EMBED, contents=label,
            config=types.EmbedContentConfig(output_dimensionality=MATCH_DIM),
        ).embeddings[0].values
        with open(f"{IMG_DIR}/{breed['image']}", 'rb') as f:
            ibytes = f.read()
        ivec = client.models.embed_content(
            model=GEMINI_EMBED,
            contents=[types.Part.from_bytes(data=ibytes, mime_type='image/png')],
            config=types.EmbedContentConfig(output_dimensionality=MATCH_DIM),
        ).embeddings[0].values
        vector_lookup.append((g, b, label))
        text_vecs.append(l2_normalize(tvec))
        image_vecs.append(l2_normalize(ivec))

text_db = np.vstack(text_vecs)
image_db = np.vstack(image_vecs)
print(f'embedded {len(vector_lookup)} breeds into {text_db.shape} text / {image_db.shape} image matrices')
```

```python
def top_matches(query_vec, db, k=5):
    sims = db @ l2_normalize(query_vec)
    idx = np.argsort(sims)[::-1][:k]
    return [(int(i), float(sims[i])) for i in idx]

def show_matches(query_gb, matches, title):
    def load(g, b):
        return PIL.Image.open(f"{IMG_DIR}/{breed_data[g]['breeds'][b]['image']}")
    fig, ax = plt.subplots(1, len(matches) + 1, figsize=(2.2 * (len(matches) + 1), 2.6))
    qg, qb = query_gb
    ax[0].imshow(load(qg, qb)); ax[0].axis('off')
    ax[0].set_title(f"query\n{breed_data[qg]['breeds'][qb]['name']}", fontsize=8)
    for col, (i, sim) in enumerate(matches, start=1):
        g, b, _ = vector_lookup[i]
        ax[col].imshow(load(g, b)); ax[col].axis('off')
        ax[col].set_title(f"{sim:.2f}\n{breed_data[g]['breeds'][b]['name']}", fontsize=8)
    fig.suptitle(title); plt.tight_layout(); plt.show()
```

### Text query -> image matches

```python
qi = 49  # Working - Siberian Husky: a visually distinctive breed that appears once in the data
q_g, q_b, q_label = vector_lookup[qi]
matches = top_matches(text_db[qi], image_db)
print(f'query text: {q_label}')
show_matches((q_g, q_b), matches, 'Text query -> Image matches')
```

### Image query -> text matches

The nearest *text* vectors to an image query -- same code, swapping which matrix is queried against.


```python
matches = top_matches(image_db[qi], text_db)
for i, sim in matches:
    print(f'{sim:.3f}  {vector_lookup[i][2]}')
```

## Batch Embeddings

For large volumes, batch jobs beat looping over online calls. On Vertex AI, `client.batches.create()` reads from Cloud Storage (JSONL) or BigQuery and writes results back to either.

> **Model support:** on Vertex AI, batch embeddings currently run on the prior text models (e.g. `text-embedding-005`), not on `gemini-embedding-2`. For `gemini-embedding-2` batch, use the Gemini Developer API's `client.batches.create_embeddings()` (Cloud Storage / uploaded-file source). Batch support is evolving -- verify current model coverage before relying on it.

> These cells create real jobs (minutes to run) and write to your bucket/dataset. `RUN_BATCH` gates them so a clean *Restart & Run All* skips them by default.


```python
RUN_BATCH = True  # <-- set True to run the batch jobs (writes to your bucket and BigQuery dataset)
bucket = gcs.bucket(BUCKET)
```

### JSONL on Cloud Storage

Write an input JSONL (one `{"content": ...}` per line), submit the job, poll to completion, then read the output shard(s).


```python
from google.genai.types import CreateBatchJobConfig

if RUN_BATCH:
    input_blob = f'{SERIES}/{EXPERIMENT}/batch_input.jsonl'
    payload = '\n'.join(json.dumps({'content': c['content']}) for c in chunks[:20])
    bucket.blob(input_blob).upload_from_string(payload, content_type='application/jsonl')

    src_uri = f'gs://{BUCKET}/{input_blob}'
    dst_uri = f'gs://{BUCKET}/{SERIES}/{EXPERIMENT}/batch_output/'
    gcs_job = client.batches.create(
        model=TEXT_EMBED, src=src_uri,
        config=CreateBatchJobConfig(dest=dst_uri),
    )
    print(f'submitted {gcs_job.name}: {gcs_job.state}')
else:
    print('Set RUN_BATCH=True to submit the Cloud Storage batch job')
```

```python
if RUN_BATCH:
    while gcs_job.state in ('JOB_STATE_PENDING', 'JOB_STATE_RUNNING'):
        time.sleep(30)
        gcs_job = client.batches.get(name=gcs_job.name)
    print(f'final state: {gcs_job.state}')

    # Read back from THIS job's own output directory. Vertex writes a timestamped
    # prediction-model-*/ subdir under the destination, so scoping to gcs_job.dest.gcs_uri
    # keeps the count correct even if the parent folder holds shards from other runs.
    out_prefix = gcs_job.dest.gcs_uri.replace(f'gs://{BUCKET}/', '')
    predictions = []
    for blob in bucket.list_blobs(prefix=out_prefix):
        if blob.name.endswith('.jsonl'):
            predictions += [json.loads(line) for line in blob.download_as_text().splitlines()]
    print(f'retrieved {len(predictions)} predictions')
```

### BigQuery in and out

Batch can read prompts from a BigQuery table and write embeddings back to another -- convenient when the corpus already lives in BigQuery.

> **Region co-location:** BigQuery batch prediction requires the source/destination dataset to live in the **same region as the Vertex job** (`LOCATION`). A dataset in the `US` multi-region will fail a `us-central1` job. The cell below creates `BQ_DATASET` in `BQ_REGION` (set to `LOCATION` in Setup) so the two always match.


```python
if RUN_BATCH:
    dataset = bigquery.Dataset(f'{PROJECT_ID}.{BQ_DATASET}')
    dataset.location = BQ_REGION
    bq.create_dataset(dataset, exists_ok=True)

    rows = [{'content': c['content']} for c in chunks[:20]]
    bq.load_table_from_json(
        rows, f'{PROJECT_ID}.{BQ_DATASET}.{BQ_TABLE}',
        job_config=bigquery.LoadJobConfig(write_disposition='WRITE_TRUNCATE', autodetect=True),
    ).result()
    print(f'loaded {len(rows)} rows to {BQ_DATASET}.{BQ_TABLE}')
```

```python
if RUN_BATCH:
    bq_src = f'bq://{PROJECT_ID}.{BQ_DATASET}.{BQ_TABLE}'
    bq_dst = f'bq://{PROJECT_ID}.{BQ_DATASET}.{BQ_TABLE}_predictions'
    bq_job = client.batches.create(
        model=TEXT_EMBED, src=bq_src,
        config=CreateBatchJobConfig(dest=bq_dst),
    )
    while bq_job.state in ('JOB_STATE_PENDING', 'JOB_STATE_RUNNING'):
        time.sleep(30)
        bq_job = client.batches.get(name=bq_job.name)
    print(f'final state: {bq_job.state}')
    if bq_job.error:  # surface why a job failed instead of silently continuing
        print(f'error: {bq_job.error}')
```

```python
if RUN_BATCH:
    out = bq.query(
        f'SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.{BQ_TABLE}_predictions` LIMIT 5'
    ).to_dataframe()
    print(f'output table columns: {list(out.columns)}')
    out.head()
```
