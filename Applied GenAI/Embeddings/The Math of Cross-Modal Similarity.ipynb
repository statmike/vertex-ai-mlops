# The Math of Cross-Modal Similarity

[`The Math of Similarity`](./The%20Math%20of%20Similarity.ipynb) showed that similarity is just geometry on vectors — and that **the same math works for any embedding**, because a vector is a vector no matter what it represents. This notebook takes the next step: what happens when the vectors come from **different modalities** — text *and* images — sharing **one** embedding space.

`gemini-embedding-2` is natively multimodal: a sentence and a photograph are mapped into the same space, so their vectors can be compared directly. That is what makes cross-modal search ("find images for this text") possible. But "same space" does not mean "same spot": the word *dog*, a photo of a dog, and a picture of the written word "dog" all land **near each other, yet not identical**. Understanding that gap — measuring it, and deciding when and how to correct for it — is the heart of using multimodal embeddings well.

**Disclaimer**

Like its companion, this notebook uses math and keeps it light and visual. Where *The Math of Similarity* used hand-made and random vectors so the geometry was easy to see, here we use **real embeddings** from `gemini-embedding-2` — so this notebook **calls the Vertex AI API** and needs a Google Cloud project with access to the model. The reasoning, though, stays in-notebook with numpy and matplotlib.

**What you will learn**

* One model can embed text and images into a **shared space** where cross-modal similarity is meaningful.
* The word "dog", a photo of a dog, and an *image of the word* "dog" are **similar but not identical** — and *why*.
* Embeddings are **invariant** to surface style (font, color, position, weight) but sensitive to content.
* The **modality gap**: text and image vectors occupy slightly separate regions — how to measure it, when it matters, and when it harmlessly cancels out.
* A toolkit for **aligning** modalities for honest mixed-media matching: mean-centering, whitening, threshold calibration, and a learned projection — with before/after retrieval quality on real data.

**TL;DR**

* Cross-modal *ranking within a fixed query→target pair* usually works with raw cosine — the modality gap is mostly a constant offset that cancels.
* When you **pool** modalities in one index, use **absolute thresholds**, or **fuse** cross-modal scores, correct the gap first. **Per-modality mean-centering** is the cheap, label-free workhorse.

````python
import io, json, os
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image, PIL.ImageDraw, PIL.ImageFont

from google import genai
from google.genai import types

client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)
print(f'google-genai client ready for {PROJECT_ID} @ {LOCATION}')
````

### Math helpers

The same tools from *The Math of Similarity*, plus small wrappers for embedding text and images with `gemini-embedding-2` and reducing vectors to 2-D with PCA for plotting (pure numpy — no extra libraries).


````python
def l2_normalize(v):
    v = np.asarray(v, dtype=float)
    n = np.linalg.norm(v)
    return v if n == 0 else v / n

def cosine(a, b):
    return float(np.dot(l2_normalize(a), l2_normalize(b)))

def embed_text(text, model=GEMINI_EMBED, dim=EMBED_DIM):
    r = client.models.embed_content(
        model=model, contents=text,
        config=types.EmbedContentConfig(output_dimensionality=dim))
    return np.array(r.embeddings[0].values)

def embed_image(image_bytes, dim=EMBED_DIM):
    r = client.models.embed_content(
        model=GEMINI_EMBED,
        contents=[types.Part.from_bytes(data=image_bytes, mime_type='image/png')],
        config=types.EmbedContentConfig(output_dimensionality=dim))
    return np.array(r.embeddings[0].values)

def pca_2d(matrix, mean=None, components=None):
    """Project rows of `matrix` to 2-D. Optionally reuse a fitted mean/components
    so multiple sets share one projection (needed to compare text vs image on one plot)."""
    X = np.asarray(matrix, dtype=float)
    if mean is None:
        mean = X.mean(axis=0)
    Xc = X - mean
    if components is None:
        _, _, vt = np.linalg.svd(Xc, full_matrices=False)
        components = vt[:2]
    return Xc @ components.T, mean, components

def show_images(imgs, captions, title=None, cols=6, size=2.3):
    """Display a row/grid of PIL images with captions -- used to make retrieval results concrete."""
    n = len(imgs); rows = (n + cols - 1) // cols
    fig, axs = plt.subplots(rows, cols, figsize=(size * cols, (size + 0.35) * rows))
    axs = np.array(axs).reshape(-1)
    for ax, im, cap in zip(axs, imgs, captions):
        ax.imshow(im); ax.axis('off'); ax.set_title(cap, fontsize=8)
    for ax in axs[len(imgs):]:
        ax.axis('off')
    if title:
        fig.suptitle(title, fontsize=13)
    plt.tight_layout(); plt.show()
````

## The Math of Semantic Embeddings (Text First)

Start on familiar ground: a dedicated **text** embedding model, `text-embedding-005`. Everything *The Math of Similarity* taught applies unchanged -- these are vectors, and cosine measures similarity. Before adding images, we explore what the text space alone encodes: similarity **structure** (a heatmap), attributes as **directions**, and even **arithmetic** on meaning. This is the baseline the multimodal model must live up to before it earns the extra modalities.


````python
concept_words = ['dog', 'puppy', 'cat', 'kitten', 'hound', 'wolf', 'coyote', 'fox', 'car', 'truck']
text005 = {w: l2_normalize(embed_text(w, model=TEXT_EMBED)) for w in concept_words}

# cosine of every word vs "dog", sorted most- to least-similar
for w in sorted(concept_words, key=lambda w: cosine(text005['dog'], text005[w]), reverse=True):
    print(f'dog vs {w:7s} -> {cosine(text005["dog"], text005[w]):.3f}')
````

**Interpretation**

The scores form a rough **gradient of relatedness** -- `puppy` right next to `dog`, animals generally above vehicles -- but read the exact order honestly: with these *bare single-word* inputs, `text-embedding-005` puts `cat` above the wild canines and can even rank `fox`/`coyote` **below** `car`. That is not a failure of the math; it is a sign that **isolated tokens are ambiguous** (a point we return to below). The reliable signal is the coarse structure, which the heatmap shows more clearly next.


### Similarity Structure At A Glance

Rather than read one column of numbers, view **every** pairwise similarity at once as a heatmap. Warm blocks are clusters of related meaning.


````python
labels = concept_words
M = np.vstack([text005[w] for w in labels])
S = M @ M.T

fig, ax = plt.subplots(figsize=(7.5, 6.5))
im = ax.imshow(S, cmap='viridis', vmin=float(S.min()), vmax=1.0)
ax.set_xticks(range(len(labels))); ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_yticks(range(len(labels))); ax.set_yticklabels(labels)
for i in range(len(labels)):
    for j in range(len(labels)):
        ax.text(j, i, f'{S[i, j]:.2f}', ha='center', va='center',
                color='white' if S[i, j] < 0.75 else 'black', fontsize=7)
fig.colorbar(im, label='cosine similarity')
ax.set_title('Pairwise cosine similarity of concept words (text-embedding-005)')
plt.tight_layout(); plt.show()
````

**Interpretation**

The heatmap makes the structure pop: the canines form a warm block, `cat`/`kitten` sit near them, and the vehicles form their own warm corner while staying cool against every animal. One picture shows what the vs-`dog` list could only reveal one row at a time.


### Attributes Are Directions

A similarity score collapses a whole relationship into one number. But specific **attributes** live along specific *directions* in the space. Pick two anchor words naming the ends of an attribute -- `domestic` and `wild` -- subtract them to get an axis, and **project** each animal onto it to score where it falls. We do the same for `small`<->`large` and `prey`<->`predator`.


````python
def axis_score(words, low_word, high_word, model=TEXT_EMBED):
    lo = l2_normalize(embed_text(low_word, model=model))
    hi = l2_normalize(embed_text(high_word, model=model))
    axis = l2_normalize(hi - lo)
    return {w: float(np.dot(l2_normalize(embed_text(w, model=model)), axis)) for w in words}

animals = ['dog', 'puppy', 'cat', 'kitten', 'hound', 'wolf', 'coyote', 'fox']
# Two axes: one that separates cleanly (domestic<->wild) and one tied to the "young"
# relationship the arithmetic section uses (young<->old). We also try a third, weaker axis
# (small<->large) on purpose -- not every anchor pair yields a clean direction.
axes_def = [('domestic', 'wild'), ('young', 'old'), ('small', 'large')]
projections = {f'{lo}  <->  {hi}': axis_score(animals, lo, hi) for lo, hi in axes_def}

fig, axs = plt.subplots(len(projections), 1, figsize=(9, 1.7 * len(projections)))
for ax, (title, proj) in zip(axs, projections.items()):
    xs = [proj[w] for w in animals]
    ax.scatter(xs, [0] * len(xs), s=50, color='#4285F4', zorder=3)
    for w in animals:
        ax.annotate(w, (proj[w], 0), rotation=40, fontsize=8, xytext=(0, 6), textcoords='offset points')
    ax.axhline(0, color='lightgrey', lw=1)
    ax.set_yticks([])
    lo_lab, hi_lab = title.split('  <->  ')
    ax.set_xlabel(f'<- {lo_lab}          {hi_lab} ->')
plt.tight_layout(); plt.show()
````

**Interpretation**

Some attributes surface as clean directions, others do not — and that contrast is the lesson:

- **`domestic ↔ wild`** separates well: `cat`/`dog` sit at the domestic end, `wolf`/`coyote`/`fox` at the wild end. The model really does encode this attribute as a measurable direction.
- **`young ↔ old`** also works: `puppy` and `kitten` land youngest — the same *young* relationship the arithmetic section exploits next.
- **`small ↔ large`** barely separates — the animals bunch near zero. The bare anchor words "small"/"large" are generic and give a noisy axis for this vocabulary.

So attributes-as-directions is real, but **axis quality depends on how well the anchor words name the attribute** — vague or ambiguous anchors give vague axes. That is a direct preview of the token-noise caution below, and a reason to choose descriptive anchors (and, later, an image-side `fluffy ↔ smooth` axis where the concept is visually crisp).


### The Math of Semantic Embeddings

If attributes are directions, we can do **arithmetic** with them. The classic example: the step from `dog` to `puppy` is an *adult→young* direction. If that direction is consistent, adding it to `cat` should land near `kitten`:

$$\text{cat} + (\text{puppy} - \text{dog}) \approx \text{kitten}$$

We compute the left side, find its nearest concept word, and separately measure whether the two *young* directions (`puppy`−`dog` and `kitten`−`cat`) are parallel.


````python
def nearest(vec, vocab, exclude=()):
    vec = l2_normalize(vec)
    scored = [(w, float(np.dot(vec, l2_normalize(text005[w])))) for w in vocab if w not in exclude]
    return sorted(scored, key=lambda x: x[1], reverse=True)

# The "young" relationship as a direction (puppy - dog), applied to cat:
young_dir = l2_normalize(text005['puppy']) - l2_normalize(text005['dog'])
analogy = l2_normalize(text005['cat']) + young_dir
print('cat + (puppy - dog)  ->  nearest concept words (excluding cat):')
for w, s in nearest(analogy, concept_words, exclude=('cat',))[:5]:
    flag = '   <-- kitten completes the analogy' if w == 'kitten' else ''
    print(f'  {w:8s} {s:.3f}{flag}')

# Are the two "young" directions parallel?
d1 = l2_normalize(text005['puppy']) - l2_normalize(text005['dog'])
d2 = l2_normalize(text005['kitten']) - l2_normalize(text005['cat'])
print(f'\ncosine( puppy-dog , kitten-cat ) = {float(np.dot(l2_normalize(d1), l2_normalize(d2))):.3f}')
print('  (near 1.0 = parallel young-directions; near 0 = unrelated)')
````

**Interpretation**

Word-vector arithmetic (`king - man + woman ~= queen`) was a hallmark of older *word* embedding models. Modern *sentence* embedding models are trained differently, so linear analogies work **less reliably** -- this cell is an honest test on our tiny vocabulary. If `kitten` tops the list (excluding `cat`) and the two *young* directions have a high cosine, the relationship really is a consistent direction here. If not, that is the lesson: rich semantic structure does **not** guarantee clean vector arithmetic. Either way, the *similarity* math the rest of this notebook relies on holds regardless.


### A Caution: Bare Tokens Are Noisy

A single word can carry **multiple senses** — `boxer` is a dog breed *and* a fighter; `fox` is an animal *and* a media company. A bare token blends these senses, so its embedding is ambiguous. Naming the intended sense in a short phrase snaps the vector to that meaning. We measure how far the same token moves when we specify each sense.


````python
# The SAME token means different things; naming the sense moves the vector a lot.
print('"boxer" -- a dog breed OR a fighter:')
print(f'  bare "boxer"                    vs "dog" = {cosine(text005["dog"], l2_normalize(embed_text("boxer", model=TEXT_EMBED))):.3f}')
print(f'  "a boxer, a breed of dog"       vs "dog" = {cosine(text005["dog"], l2_normalize(embed_text("a boxer, a breed of dog", model=TEXT_EMBED))):.3f}')
print(f'  "a boxer, a professional fighter" vs "dog" = {cosine(text005["dog"], l2_normalize(embed_text("a boxer, a professional fighter", model=TEXT_EMBED))):.3f}')

wild = l2_normalize(embed_text('a wild animal', model=TEXT_EMBED))
print('\n"fox" -- an animal OR a media company:')
print(f'  bare "fox"                      vs "a wild animal" = {cosine(wild, l2_normalize(embed_text("fox", model=TEXT_EMBED))):.3f}')
print(f'  "a fox, a wild animal"          vs "a wild animal" = {cosine(wild, l2_normalize(embed_text("a fox, a wild animal", model=TEXT_EMBED))):.3f}')
print(f'  "Fox, the media company"        vs "a wild animal" = {cosine(wild, l2_normalize(embed_text("Fox, the media company", model=TEXT_EMBED))):.3f}')
````

**Interpretation**

The same token swings sharply with context: "a boxer, a breed of dog" sits much closer to `dog` than "a boxer, a professional fighter", and "a fox, a wild animal" is far more animal-like than "Fox, the media company". The bare word sits in between — a blur of its senses.

This is exactly why the multimodal experiments below embed **full breed descriptions**, not lone words, and it is a good habit for any retrieval system: **embed a phrase or sentence that pins down the meaning**, not a bare keyword, whenever you can. (It is also why the `small ↔ large` axis above was weak — generic anchor words are themselves under-specified.)

## Switch To `gemini-embedding-2` — Redo The Text

Now the same words through the natively-multimodal model. If it is going to embed images into the *same* space as text, it must first handle text at least as well. We redo the identical comparison.


````python
text_g2 = {w: l2_normalize(embed_text(w)) for w in concept_words}

# Compare the two models side by side, ordered by the gemini-2 similarity to "dog".
order = sorted(concept_words, key=lambda w: cosine(text_g2['dog'], text_g2[w]), reverse=True)
print(f"{'word':8s}{'005':>8s}{'g2':>8s}{'g2 rank':>9s}")
for rank, w in enumerate(order):
    print(f'{w:8s}{cosine(text005["dog"], text005[w]):>8.3f}{cosine(text_g2["dog"], text_g2[w]):>8.3f}{rank:>9d}')
````

**Interpretation**

`gemini-embedding-2` keeps the big picture — `puppy` on top, vehicles near the bottom — so it clearly handles text. But look closely and the two models are **not identical**: gemini-2 sits on a **higher, more compressed scale** (its similarities bunch into a narrower high band), and the *middle* of the ordering can shuffle — a bare token like `wolf` or `fox` may land differently than in `text-embedding-005`, because short out-of-context words give each model little to work with and the models learned different neighborhoods.

Two takeaways, both important going forward:
- **Compare scores only within one model and one modality-pair**, never across models — the absolute numbers live on different scales.
- A single word is a *thin* input; the richer text used later (full breed descriptions) gives far more stable structure. With text trust established, we can bring in images.

## The "dog" Probe: Similar But Not Identical

Here is the core experiment. We take one concept — **dog** — and represent it four ways, then measure how the vectors relate:

1. the **word** `dog` (text)
2. a **photo** of a dog (image)
3. an **image of the written word** "dog" (text rendered as a picture)
4. **styled variations** of that word-image (font, color, position, weight)

All four are embedded with `gemini-embedding-2` into the same space. The question is not *are they identical* (they will not be) but *how do they relate, and what does the model treat as the same*.


### Build the stimuli

We render the word-images with PIL (no image-generation API needed) and reuse a real dog **photo** from the sibling [Generate](../Generate) folder. Styled variations change only surface appearance — the *content* is always the word "dog".


````python
# Load breed data (name, descriptions, image filename per breed) and pick a hero dog photo.
with open(f'{IMG_DIR}/breed-data.jsonl') as f:
    breed_data = [json.loads(line) for line in f]

def breed_image_bytes(image_name):
    with open(f'{IMG_DIR}/{image_name}', 'rb') as f:
        return f.read()

# Hero photo: first breed's image.
hero = breed_data[0]['breeds'][0]
hero_photo_bytes = breed_image_bytes(hero['image'])
print(f"hero dog photo: {breed_data[0]['name']} - {hero['name']} ({hero['image']})")
````

````python
# Render text as a PNG image with controllable style. Returns raw PNG bytes.
FONT_PATHS = {
    'sans':  '/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf',
    'serif': '/usr/share/fonts/truetype/dejavu/DejaVuSerif.ttf',
    'mono':  '/usr/share/fonts/truetype/dejavu/DejaVuSansMono.ttf',
    'bold':  '/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf',
}

def render_word(word, font='sans', size=96, color=(0, 0, 0), bg=(255, 255, 255),
                position='center', canvas=(400, 200)):
    img = PIL.Image.new('RGB', canvas, bg)
    draw = PIL.ImageDraw.Draw(img)
    try:
        fnt = PIL.ImageFont.truetype(FONT_PATHS.get(font, FONT_PATHS['sans']), size)
    except Exception:
        fnt = PIL.ImageFont.load_default()
    tb = draw.textbbox((0, 0), word, font=fnt)
    tw, th = tb[2] - tb[0], tb[3] - tb[1]
    if position == 'center':
        xy = ((canvas[0] - tw) / 2 - tb[0], (canvas[1] - th) / 2 - tb[1])
    elif position == 'top-left':
        xy = (10, 10)
    elif position == 'bottom-right':
        xy = (canvas[0] - tw - 10 - tb[0], canvas[1] - th - 10 - tb[1])
    else:
        xy = ((canvas[0] - tw) / 2 - tb[0], (canvas[1] - th) / 2 - tb[1])
    draw.text(xy, word, font=fnt, fill=color)
    buf = io.BytesIO()
    img.save(buf, format='PNG')
    return buf.getvalue()

# Baseline word-image plus four styled variations (each changes only surface appearance).
word_image_bytes = render_word('dog')
variations = {
    'baseline':      render_word('dog'),
    'serif font':    render_word('dog', font='serif'),
    'red color':     render_word('dog', color=(200, 30, 30)),
    'top-left':      render_word('dog', position='top-left'),
    'bold + large':  render_word('dog', font='bold', size=140),
}
print(f'rendered {len(variations)} styled word-images + 1 real photo')
````

### Measurement 1 — word vs photo (cross-modal semantic alignment)

Does the **word** "dog" (text) land near a **photo** of a dog (image)? This is the fundamental cross-modal question. We compare it against an unrelated photo (a vehicle-free control from another breed group is still a dog, so we also embed the word "car" as a semantic control).


````python
v_word_dog = l2_normalize(embed_text('dog'))
v_photo_dog = l2_normalize(embed_image(hero_photo_bytes))
v_word_car = l2_normalize(embed_text('car'))

print(f'cosine(word "dog",  photo of dog)  = {cosine(v_word_dog, v_photo_dog):.3f}')
print(f'cosine(word "car",  photo of dog)  = {cosine(v_word_car, v_photo_dog):.3f}')
print(f'cosine(word "dog",  word "car")    = {cosine(v_word_dog, v_word_car):.3f}')
````

**Interpretation**

If the shared space works, `word "dog"` ↔ `photo of dog` scores clearly higher than `word "car"` ↔ `photo of dog` — cross-modal similarity tracks *meaning*, not modality. Note the cross-modal score will likely be **lower** than a within-text score like `dog`↔`puppy` even when the match is correct: that is the **modality gap**, quantified in the next section. Similar, not identical.


### Measurement 2 — photo vs image-of-the-word (does it read text in images?)

Now compare the **photo of a dog** to an **image of the written word "dog"**. Both are images, but one *depicts* a dog and the other *spells* it. Where does the word-image land — nearer the animal photo, or nearer the text token "dog"?


````python
v_wordimg_dog = l2_normalize(embed_image(word_image_bytes))

print(f'cosine(image-of-word "dog", word "dog" [text]) = {cosine(v_wordimg_dog, v_word_dog):.3f}')
print(f'cosine(image-of-word "dog", photo of a dog)    = {cosine(v_wordimg_dog, v_photo_dog):.3f}')
print(f'cosine(photo of a dog,      word "dog" [text]) = {cosine(v_photo_dog, v_word_dog):.3f}')
````

**Interpretation**

Two signals are competing here — *modality* (it is an image) and *content* (it spells "dog") — and the result shows **modality wins**: the image-of-the-word lands closest to the actual dog **photo** (both are images, and image↔image similarity runs high, as we measure next), and only modestly close to the **text** token "dog". That said, there *is* a text-reading signal: the image-of-the-word sits a bit nearer the text token than the plain dog-photo does, so the model reads the letters somewhat — it just weighs "this is an image" more heavily than "this image says dog."

The practical lesson: a picture of text is treated first as a picture. If you need the *words inside images* to drive retrieval, don't rely on the raw image embedding — extract the text (OCR) and embed that as text, or embed both and combine them.


### Measurement 3 — style variations (invariance)

The five word-images all say "dog" but differ in font, color, position, and weight. If the model encodes **content over surface style**, these should cluster tightly. The spread among them *quantifies* how much the model ignores appearance.


````python
var_vecs = {name: l2_normalize(embed_image(b)) for name, b in variations.items()}
names = list(var_vecs)
base = var_vecs['baseline']

print('cosine of each styled word-image vs the baseline word-image:')
for name in names:
    print(f'  {name:14s} -> {cosine(base, var_vecs[name]):.3f}')

# Pairwise spread: how tightly do all variations cluster?
mat = np.vstack([var_vecs[n] for n in names])
sims = mat @ mat.T
off = sims[np.triu_indices(len(names), k=1)]
print(f'\nmean pairwise cosine among variations = {off.mean():.3f}  (min {off.min():.3f})')
````

**Interpretation**

High cosines across all variations (mean near 1.0) mean the model treats font/color/position/weight as **irrelevant surface detail** — it embeds *what the word says*, not *how it looks*. This is the image-side analogue of semantic invariance in text, and it is exactly what makes embeddings robust for retrieval: two photos of the same concept, or two renderings of the same word, land in the same place.

Three representations of one concept, similar but not identical. That separation is the modality gap, which we now measure properly.

## The Modality Gap

The probe hinted that text and image vectors of the same concept do not coincide. Across *many* concepts this becomes a clear, systematic pattern: **text embeddings cluster in one region of the space and image embeddings in another**, even for matched pairs. This is the well-known *modality gap* seen in contrastively-trained multimodal models. We measure it on the breed dataset, which gives many text↔image pairs.


````python
# Build matched pairs: each breed's text (name + description) and its photo.
# Sample ACROSS breed groups (not just the first N) so text/image stats and the
# label-hungry aligners (whitening, Procrustes) see diverse, representative data.
N_BREEDS = 80   # raise toward the full ~118 for a denser study; lower to run faster
all_breeds = [(f"{b['name']}. {b['physical_description']}", b['image'], g['name'], b['name'])
              for g in breed_data for b in g['breeds']]
step = max(1, len(all_breeds) // N_BREEDS)
pairs = all_breeds[::step][:N_BREEDS]

text_vecs = np.vstack([l2_normalize(embed_text(t)) for t, _, _, _ in pairs])
image_vecs = np.vstack([l2_normalize(embed_image(breed_image_bytes(img))) for _, img, _, _ in pairs])
print(f'embedded {len(pairs)} matched text/image breed pairs across {len(breed_data)} groups')
````

### See The Data

Before measuring the gap, look at what we embed: each breed contributes a **photo** (its image vector) and a **text description** (its text vector). A few examples:


````python
sample = pairs[:6]
imgs = [PIL.Image.open(io.BytesIO(breed_image_bytes(img))) for _, img, _, _ in sample]
caps = [b for _, _, g, b in sample]
show_images(imgs, caps, title='Sample breed pairs: the photo (image) + the breed text we embed', cols=6)
````

````python
# Within-modality vs across-modality mean cosine.
def mean_offdiag(M):
    s = M @ M.T
    return s[np.triu_indices(len(M), k=1)].mean()

within_text = mean_offdiag(text_vecs)
within_image = mean_offdiag(image_vecs)
# across: matched pairs (diagonal) vs mismatched (off-diagonal)
cross = text_vecs @ image_vecs.T
matched = np.diag(cross).mean()
mismatched = cross[~np.eye(len(pairs), dtype=bool)].mean()

print(f'mean cosine within text   (diff breeds) = {within_text:.3f}')
print(f'mean cosine within images (diff breeds) = {within_image:.3f}')
print(f'mean cosine across, MATCHED   pairs     = {matched:.3f}')
print(f'mean cosine across, MISMATCHED pairs    = {mismatched:.3f}')
````

**Interpretation**

Two things to look for. **(1) The gap:** matched cross-modal pairs typically score *lower* than within-modality pairs — text-to-text and image-to-image similarities live on a higher scale than text-to-image, because each modality occupies its own region. **(2) Signal survives anyway:** matched pairs should still beat mismatched pairs, so cross-modal retrieval *works* — the right image is still the nearest, just at a compressed absolute score.


````python
# Visualize the two cones: project text + image vectors into one shared PCA plane.
allv = np.vstack([text_vecs, image_vecs])
coords, mean, comps = pca_2d(allv)
tc, ic = coords[:len(pairs)], coords[len(pairs):]

fig, ax = plt.subplots(figsize=(9, 7))
ax.scatter(tc[:, 0], tc[:, 1], c='#4285F4', s=60, label='text (breed descriptions)')
ax.scatter(ic[:, 0], ic[:, 1], c='#EA4335', s=60, label='image (breed photos)')
# draw the centroids and the offset between them
tcen, icen = tc.mean(0), ic.mean(0)
ax.scatter(*tcen, c='#4285F4', s=300, marker='*', edgecolor='k', zorder=5)
ax.scatter(*icen, c='#EA4335', s=300, marker='*', edgecolor='k', zorder=5)
ax.annotate('', xy=icen, xytext=tcen, arrowprops=dict(arrowstyle='->', lw=2, color='black'))
ax.set_xlabel('PC 1'); ax.set_ylabel('PC 2')
ax.set_title('The modality gap: text and image vectors occupy separate regions')
ax.legend(); plt.tight_layout(); plt.show()
````

**Interpretation**

The two clouds (stars = centroids) are offset — that arrow between centroids **is** the modality gap. Crucially it looks close to a single shared **translation**: image vectors are roughly the text vectors shifted by a constant. That observation is the key to the next point.


### The Headline: For Ranking, The Gap Cancels

This is the single most useful fact about cross-modal search, so make it the takeaway: **if the gap is roughly a constant offset $g$ (image centroid - text centroid), it does not change *rankings*.** For a text query, every image candidate's score shifts by about the same amount, so the *order* -- and therefore top-k retrieval -- is nearly untouched. We test it directly: does subtracting the gap change *who* the top matches are?


````python
gap = image_vecs.mean(0) - text_vecs.mean(0)

def topk_for_query(qi, db, k=5):
    q = text_vecs[qi]
    return np.argsort(db @ q)[::-1][:k]

changed = 0
for qi in range(len(pairs)):
    raw = topk_for_query(qi, image_vecs)
    shifted = topk_for_query(qi, image_vecs - gap)  # remove the constant offset
    if list(raw) != list(shifted):
        changed += 1
print(f'queries whose top-5 image matches changed after removing the gap: {changed}/{len(pairs)}')
print('=> if this is small, the gap is ~a constant offset that does not affect ranking')
````

**Interpretation -- the practical headline**

**If that count is small (near zero), pure cross-modal retrieval needs no correction at all.** The modality gap is real but, for *ranking within a fixed query->target direction*, it cancels out. Reach for the alignment toolkit below only when absolute scores must be comparable **across** modalities -- pooling into one index, absolute thresholds, or score fusion. The next section shows one of those cases breaking, then fixes it.


---
## Aligning Modalities For Mixed-Media Matching

The gap bites when you: **pool** text and image results in one ranked list; apply an **absolute threshold** ("keep matches above 0.5"); or **fuse** cross-modal scores. Then text-to-text similarities (naturally higher) drown out cross-modal ones. This section works through four alignment methods, simplest to most powerful, each **evaluated on a real task**: use a text query to retrieve the matching breed *image*, where the correct answer is known.


### First, Watch The Gap Break A Pooled Search

The clearest way to see *when* alignment matters: put **text and image vectors in one shared index** and run a single text query against all of them. Because text<->text similarities sit on a higher scale than text<->image, the same-modality results dominate and the cross-modal match we actually want gets buried.


````python
qi = 0
q = text_vecs[qi]
pooled = np.vstack([text_vecs, image_vecs])
modality = ['text'] * len(pairs) + ['image'] * len(pairs)
is_correct_image = [False] * len(pairs) + [j == qi for j in range(len(pairs))]

scores = pooled @ q
ranked = [o for o in np.argsort(scores)[::-1] if not (modality[o] == 'text' and o == qi)]
print(f'Pooled search (text + image in ONE index), query = text for "{pairs[qi][3]}":')
n_img_top = 0
for o in ranked[:10]:
    tag = '   <-- the CORRECT image' if is_correct_image[o] else ''
    n_img_top += modality[o] == 'image'
    print(f'  {modality[o]:5s}  {scores[o]:.3f}  {pairs[o % len(pairs)][3]}{tag}')
img_rank = ranked.index(len(pairs) + qi) + 1
print(f'\nimages in the top 10: {n_img_top}/10   |   the correct image ranks #{img_rank} overall')
````

**Interpretation**

The top of the pooled list is almost all **text** -- the correct *image* is pushed far down, even though it is a perfect content match. This is the gap doing real damage: the moment modalities share a ranked list, the higher within-modality scale wins. Ranking *within* a single query->target pair was fine (previous section); **pooling across modalities is where it breaks.**


### The evaluation task

For each breed, the text description is the query and the goal is to retrieve that breed's photo. Because matched pairs share an index, **Recall@1** = fraction of queries whose nearest image is the correct one, and mean reciprocal rank (MRR) summarizes ranking quality. We score the raw (unaligned) case first as a baseline.


````python
def retrieval_metrics(query_vecs, db_vecs):
    """query i's correct answer is db row i. Returns (recall@1, MRR)."""
    sims = query_vecs @ db_vecs.T
    ranks = []
    for i in range(len(query_vecs)):
        order = np.argsort(sims[i])[::-1]
        rank = int(np.where(order == i)[0][0]) + 1
        ranks.append(rank)
    ranks = np.array(ranks)
    return (ranks == 1).mean(), (1.0 / ranks).mean()

r1, mrr = retrieval_metrics(text_vecs, image_vecs)
print(f'RAW (no alignment):  Recall@1 = {r1:.3f}   MRR = {mrr:.3f}')
````

### Method 1 — Per-modality mean-centering (the workhorse)

Subtract each modality's own mean vector, then re-normalize. This removes the constant offset that *is* most of the gap — cheap, needs **no labels**, and usually the biggest win per unit effort.

$$\hat{t}_i = \frac{t_i - \bar{t}}{||t_i - \bar{t}||}, \qquad \hat{m}_i = \frac{m_i - \bar{m}}{||m_i - \bar{m}||}$$


````python
def center_normalize(vecs, mean=None):
    mean = vecs.mean(0) if mean is None else mean
    out = vecs - mean
    out = out / np.linalg.norm(out, axis=1, keepdims=True)
    return out, mean

text_c, _ = center_normalize(text_vecs)
image_c, _ = center_normalize(image_vecs)
r1_c, mrr_c = retrieval_metrics(text_c, image_c)
print(f'MEAN-CENTERED:       Recall@1 = {r1_c:.3f}   MRR = {mrr_c:.3f}')
print(f'   (raw was         Recall@1 = {r1:.3f}   MRR = {mrr:.3f})')
````

**Interpretation**

Mean-centering leaves *within-pair ranking* largely unchanged (consistent with the offset-cancels result) but makes the two modalities **share a common origin**, so their scores become comparable — which is what pooling and thresholding need. Any change in Recall@1 here reflects second-order gap structure beyond a pure constant.


### Revisit: Mean-Centering Fixes The Pooled Search

We just saw pooling break. Now re-run the identical pooled query using the **mean-centered** vectors from Method 1 -- subtracting each modality's own mean puts both on a common origin, so text and image scores finally compete fairly.


````python
tc_all, _ = center_normalize(text_vecs)
ic_all, _ = center_normalize(image_vecs)
q_c = tc_all[qi]
pooled_c = np.vstack([tc_all, ic_all])
scores_c = pooled_c @ q_c
ranked_c = [o for o in np.argsort(scores_c)[::-1] if not (modality[o] == 'text' and o == qi)]
print('Same pooled query, AFTER per-modality mean-centering:')
n_img_top = 0
for o in ranked_c[:10]:
    tag = '   <-- the CORRECT image' if is_correct_image[o] else ''
    n_img_top += modality[o] == 'image'
    print(f'  {modality[o]:5s}  {scores_c[o]:.3f}  {pairs[o % len(pairs)][3]}{tag}')
img_rank_c = ranked_c.index(len(pairs) + qi) + 1
print(f'\nimages in the top 10: {n_img_top}/10   |   the correct image now ranks #{img_rank_c} overall')
````

**Interpretation**

With both modalities centered, the correct image climbs and text no longer dominates by default -- the same query that failed now works. This is the concrete payoff: **mean-centering is what makes pooled, cross-modal ranked lists honest**, and it needed no labels to do it.


### Method 2 — Whitening (ZCA)

Centering removes the offset; **whitening** also removes each modality's *shape* — it decorrelates the dimensions and scales them to unit variance, making the cloud isotropic (spherical). Use it when the modalities differ in more than just location. We fit the transform on each modality and apply it, then re-normalize.

$$W = \Sigma^{-1/2}, \qquad \hat{x} = W\,(x - \bar{x})$$


````python
def whitener(vecs, eps=1e-5):
    mean = vecs.mean(0)
    Xc = vecs - mean
    cov = np.cov(Xc, rowvar=False)
    U, S, _ = np.linalg.svd(cov)
    W = U @ np.diag(1.0 / np.sqrt(S + eps)) @ U.T   # ZCA whitening matrix
    return mean, W

def apply_whiten(vecs, mean, W):
    out = (vecs - mean) @ W.T
    return out / np.linalg.norm(out, axis=1, keepdims=True)

tm, tW = whitener(text_vecs)
im, iW = whitener(image_vecs)
text_w = apply_whiten(text_vecs, tm, tW)
image_w = apply_whiten(image_vecs, im, iW)
r1_w, mrr_w = retrieval_metrics(text_w, image_w)
print(f'WHITENED (ZCA):      Recall@1 = {r1_w:.3f}   MRR = {mrr_w:.3f}')
````

**Interpretation**

Whitening is more aggressive than centering. It can help when one modality's variance is concentrated in a few directions, but with limited data the covariance estimate is noisy and it can *hurt* — a real tradeoff the numbers here make concrete. Note it needs enough samples per modality to estimate the covariance well (we use `N_BREEDS` vectors in `EMBED_DIM` dimensions — deliberately watch for over-fitting).


### Method 3 — Threshold calibration (leave the vectors alone)

Sometimes you cannot modify stored vectors (they live in a served index). Instead, calibrate the *scores*: standardize each modality-pair's similarity distribution (subtract its mean, divide by its std) so a "0.6" means the same thing whether it came from text-to-text or text-to-image. This changes comparability without touching a single embedding.


````python
# Calibrate cross-modal scores against the text->image score distribution.
cross_raw = text_vecs @ image_vecs.T
mu, sd = cross_raw.mean(), cross_raw.std()
cross_z = (cross_raw - mu) / sd   # z-scored similarities

# ranking within the pair is unchanged by an affine transform, so Recall@1/MRR match raw;
# the value is a COMPARABLE scale. Show a threshold example:
thresh = 2.0  # keep matches > 2 std above the mean cross-modal score
kept = (cross_z > thresh)
print(f'z-scored cross-modal scores: mean {cross_z.mean():.2f}, std {cross_z.std():.2f}')
print(f'matched pairs kept at z>{thresh}: {np.diag(kept).mean():.2%}')
print(f'mismatched pairs kept at z>{thresh}: {kept[~np.eye(len(pairs),dtype=bool)].mean():.2%}')
````

**Interpretation**

Calibration does not change *ranking* within a modality-pair (an affine transform preserves order, so Recall@1/MRR are identical to raw) — its job is to put different pairings on a **common, interpretable scale** so a single absolute threshold behaves sensibly across them. This is the right tool when the vectors are frozen but you need comparable scores or a shared cutoff.


### Method 4 — Learned projection (Procrustes)

With **matched pairs as labels**, fit a mapping that pulls one modality onto the other. The orthogonal **Procrustes** solution finds the rotation $R$ minimizing $||T R - M||$ (map text onto image space) — a true learned alignment, at the cost of needing paired training data. We fit on a train split and evaluate on a held-out test split to keep it honest.


````python
rng = np.random.default_rng(0)
idx = rng.permutation(len(pairs))
split = len(pairs) // 2
train, test = idx[:split], idx[split:]

# Orthogonal Procrustes: R = argmin ||T R - M|| over orthogonal R  =>  R = U V^T from SVD(T^T M)
T_tr, M_tr = text_vecs[train], image_vecs[train]
U, _, Vt = np.linalg.svd(T_tr.T @ M_tr)
R = U @ Vt

def proj_norm(X):
    Y = X @ R
    return Y / np.linalg.norm(Y, axis=1, keepdims=True)

# Evaluate on held-out test pairs (re-index so correct answer is the diagonal of the test block).
text_te = proj_norm(text_vecs[test])
image_te = image_vecs[test] / np.linalg.norm(image_vecs[test], axis=1, keepdims=True)
r1_p, mrr_p = retrieval_metrics(text_te, image_te)

# raw baseline on the same test split for a fair comparison
r1_te_raw, mrr_te_raw = retrieval_metrics(
    text_vecs[test] / np.linalg.norm(text_vecs[test], axis=1, keepdims=True), image_te)
print(f'PROCRUSTES (test split): Recall@1 = {r1_p:.3f}   MRR = {mrr_p:.3f}')
print(f'raw       (test split):  Recall@1 = {r1_te_raw:.3f}   MRR = {mrr_te_raw:.3f}')
````

**Interpretation**

Procrustes can be the strongest aligner *when you have reliable matched pairs* — it learns an actual rotation between the spaces. The catch is exactly that requirement: it needs labeled pairs and enough of them to generalize (hence the train/test split — judge it on the held-out numbers, not the fit). With few pairs it may not beat label-free centering.


### Comparison

Put the methods side by side — the same tradeoff table style as *The Math of Similarity*'s conclusion.


````python
# Two groups so the comparison is fair: label-free methods on the full set, and the
# label-dependent Procrustes against its OWN raw baseline on the identical held-out split.
print('Label-free methods (evaluated on the full set):')
print(f"  {'method':16s}{'Recall@1':>10s}{'MRR':>8s}   when to reach for it")
full = [
    ('raw (none)',     r1,   mrr,   'nothing to do / pure ranking'),
    ('mean-centering', r1_c, mrr_c, 'pooling / thresholding across modalities'),
    ('whitening',      r1_w, mrr_w, 'modalities differ in shape, not just offset'),
]
for name, a, b, when in full:
    print(f'  {name:16s}{a:>10.3f}{b:>8.3f}   {when}')

print('\nLearned projection (needs labels; judged on a held-out split vs raw on the SAME split):')
print(f"  {'method':16s}{'Recall@1':>10s}{'MRR':>8s}")
for name, a, b in [('raw (test)', r1_te_raw, mrr_te_raw), ('procrustes (test)', r1_p, mrr_p)]:
    print(f'  {name:16s}{a:>10.3f}{b:>8.3f}')
print('\n=> here Procrustes UNDER-performs its own raw baseline: with few pairs the learned')
print('   rotation overfits the train split and fails to generalize (see interpretation).')
````

| Method | Removes | Needs labels? | Cost | Best when |
|---|---|---|---|---|
| **Mean-centering** | constant offset | No | trivial | pooling / thresholds across modalities — the default first move |
| **Whitening (ZCA)** | offset + per-modality shape | No | moderate; needs samples | modalities differ in variance structure, not just location |
| **Threshold calibration** | nothing (rescales scores) | No | trivial | vectors are frozen; you need a comparable scale or shared cutoff |
| **Learned projection (Procrustes)** | offset + rotation | **Yes** | fit + labels | you have reliable matched pairs **and enough of them** |

**What this run shows:** the label-free methods barely move Recall@1 (the gap is mostly a constant offset, so ranking was already fine) — their real value is making scores *comparable* for pooling and thresholds. Procrustes actually **hurt** here: with only a handful of training pairs the learned rotation overfits and loses to plain raw cosine on held-out data. That is not a knock on the method — it is the honest lesson that **a learned aligner needs enough reliable pairs to earn its keep**, and below that bar the free methods win.

**The practical answer:** for pure cross-modal *ranking*, you often need nothing — the gap cancels. The moment you mix modalities in one index or use absolute thresholds, reach for **mean-centering** first; escalate to whitening or a learned projection only when you have the data and the evaluation says it helps.


### A Decision Guide

```
Comparing within ONE query->target direction? (e.g. text query -> image results)
   |
   +-- YES: rank by raw cosine. The gap cancels; do nothing.   <-- most retrieval
   |
   +-- NO: pooling modalities, using ABSOLUTE thresholds, or FUSING scores?
             |
             +-- Start with PER-MODALITY MEAN-CENTERING  (free, no labels)
             +-- Modalities differ in shape?            -> WHITENING
             +-- Vectors frozen in a served index?       -> THRESHOLD CALIBRATION
             +-- Many reliable matched pairs?            -> LEARNED PROJECTION
```

**In one line:** default to raw cosine for ranking; the moment scores must be comparable *across* modalities, mean-center first and escalate only if your evaluation says so.


### Bonus: a characteristic axis ("fluffy" vs "smooth")

Cross-modal matching is not limited to identity. The breed descriptions carry **coat texture** ("fluffy", "smooth", "silky", "wiry"), so a text query for a characteristic should retrieve breed *images* sharing it — if the image embeddings capture texture at all. This is an honest probe of how much visual detail survives into the vectors.


````python
for query in ['a fluffy dog', 'a smooth short-haired dog']:
    q = l2_normalize(embed_text(query))
    order = np.argsort(image_vecs @ q)[::-1][:5]
    imgs = [PIL.Image.open(io.BytesIO(breed_image_bytes(pairs[j][1]))) for j in order]
    caps = [f'{pairs[j][3]}\n{float(image_vecs[j] @ q):.2f}' for j in order]
    show_images(imgs, caps, title=f'text query: "{query}"  ->  top image matches', cols=5)
````

**Interpretation**

If the top image matches for "fluffy" skew toward visibly fluffy breeds (and "smooth" toward short-coated ones), the image vectors encode **characteristics**, not just breed identity — cross-modal search can then answer descriptive queries. If not, it reveals a limit of what the image embeddings captured from these generated pictures. Either way, it is a measurement, not an assumption.
