# Retrieval - Vertex AI Vector Search 2

In prior workflows, a series of documents was [processed into chunks](../../Chunking/readme.md), and for each chunk, [embeddings](../../Embeddings/readme.md) were generated. The goal is to set up a retrieval system that can be queried for matching content based on similarity to a query. This workflow implements retrieval using [Vertex AI Vector Search 2](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/vector-search-2/overview) — a fully managed, serverless vector database.

**What makes Vector Search 2 different?**

Vector Search 2 unifies data and vectors in a single **Collection**. Unlike [Vector Search 1.0](../Retrieval%20-%20Vertex%20AI%20Vector%20Search.ipynb), which requires deploying indexes to endpoints and managing embeddings separately, Vector Search 2 stores everything together — vectors, text, and metadata — in **DataObjects**. It also supports **auto-embedding**, where the service generates embeddings at import and query time using a configured Vertex AI model, eliminating the need for a client-side embedding pipeline.

This workflow demonstrates:
- **Collection creation** with both BYOE (bring your own embedding) and auto-embedding vector fields
- **Vector search** — query with a pre-computed embedding vector
- **Semantic search** — query with natural language text, auto-embedded by VS2
- **Text search** — full-text keyword search with phrase and Boolean operators
- **Pre-filtered search** — scoped similarity search using MongoDB-style filters
- **Query** — structured filter-only retrieval and aggregation
- **Hybrid search** — combine vector, semantic, and text searches with Reciprocal Rank Fusion (RRF)
- **VertexRanker reranking** — server-side semantic reranking after hybrid fusion (documented but not yet in the Python SDK)
- **Data object management** — get, update, delete individual DataObjects
- **RAG** — a simple Retrieval Augmented Generation example
- **Performance profiling** — sequential retrieval latency measurements

> **ANN Indexing:** For approximate nearest neighbor indexing, recall calibration, and ANN-accelerated search, see the companion notebook [Retrieval - Vertex AI Vector Search 2 — ANN Indexes](Retrieval%20-%20Vertex%20AI%20Vector%20Search%202%20-%20ANN%20Indexes.ipynb). All searches in this notebook use brute-force kNN (exact nearest neighbor).

## Create Collection

A [Collection](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/vector-search-2/overview) defines the schema for your data and vectors. It has two schemas:
- **`data_schema`** — JSON Schema defining the structured fields each DataObject stores
- **`vector_schema`** — defines embedding fields, either manual (BYOE) or auto-generated

This collection demonstrates both approaches side by side:
- `chunk_embedding` — a BYOE dense vector field where we provide the pre-computed 768-dimensional embeddings from `text-embedding-004`
- `auto_embedding` — an auto-generated dense vector field where VS2 generates embeddings using `EMBEDDING_MODEL` from the `content` field at import time, at the model's native 3072 dimensions

```python
data_schema = {
    'type': 'object',
    'properties': {
        'gse': {'type': 'string'},
        'chunk_id': {'type': 'string'},
        'content': {'type': 'string'},
    },
}

vector_schema = {
    'chunk_embedding': {
        'dense_vector': {
            'dimensions': 768,
        },
    },
    'auto_embedding': {
        'dense_vector': {
            'dimensions': 3072,
            'vertex_embedding_config': {
                'model_id': EMBEDDING_MODEL,
                'text_template': '{content}',
                'task_type': 'RETRIEVAL_DOCUMENT',
            },
        },
    },
}
```

```python
from google.api_core.exceptions import AlreadyExists, ResourceExhausted

try:
    operation = vs_client.create_collection(
        vs.CreateCollectionRequest(
            parent=parent,
            collection_id=COLLECTION_ID,
            collection={'data_schema': data_schema, 'vector_schema': vector_schema},
        )
    )
    print('Creating collection...')
    result = operation.result()
    print(f'Created: {result.name}')
except AlreadyExists:
    print(f'Collection already exists: {collection_parent}')
```

Verify the collection:

```python
collection = vs_client.get_collection(vs.GetCollectionRequest(name=collection_parent))
print(f'Name: {collection.name}')
print(f'Data schema properties: {list(collection.data_schema.get("properties", {}).keys())}')
print(f'Vector schema fields: {list(collection.vector_schema.keys())}')
```

## Create Data Objects

[DataObjects](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/vector-search-2/data-objects/data-objects) are individual JSON items stored in a Collection. Each has:
- A unique `data_object_id` (we use the `chunk_id`)
- A `data` dict with the structured fields
- A `vectors` dict with embedding values

We provide the pre-computed embedding for `chunk_embedding` (BYOE). The `auto_embedding` field is left empty — VS2 will auto-generate it from the `content` field using the configured `EMBEDDING_MODEL`.

DataObjects are created in batches of 100 using `batch_create_data_objects()`. Because VS2 auto-generates embeddings for each DataObject at import time, large imports can exceed the embedding model's online prediction quota. The loop below retries with exponential backoff on `ResourceExhausted` (429) errors to pace through the quota.

```python
BATCH_SIZE = 100

requests = []
for chunk in content_chunks:
    requests.append(
        vs.CreateDataObjectRequest(
            parent=collection_parent,
            data_object_id=chunk['chunk_id'],
            data_object={
                'data': {
                    'gse': chunk['gse'],
                    'chunk_id': chunk['chunk_id'],
                    'content': chunk['content'],
                },
                'vectors': {
                    'chunk_embedding': {
                        'dense': {'values': chunk['embedding']},
                    },
                },
            },
        )
    )

print(f'Prepared {len(requests)} DataObject requests')
```

```python
created = 0
skipped = 0

for i in range(0, len(requests), BATCH_SIZE):
    batch = requests[i:i + BATCH_SIZE]

    for attempt in range(8):
        try:
            do_client.batch_create_data_objects(
                vs.BatchCreateDataObjectsRequest(
                    parent=collection_parent,
                    requests=batch,
                )
            )
            created += len(batch)
            break
        except AlreadyExists:
            skipped += len(batch)
            break
        except ResourceExhausted:
            wait = min(2 ** attempt * 5, 300)
            print(f'  Quota limited — retrying in {wait}s (attempt {attempt + 1}/8)')
            time.sleep(wait)
    else:
        print(f'  FAILED: batch at index {i} after 8 retries')

    batch_num = i // BATCH_SIZE + 1
    if batch_num % 10 == 0 or batch_num == 1:
        print(f'  Progress: {created + skipped}/{len(requests)} (created: {created}, skipped: {skipped})')

print(f'\nDone: {created} created, {skipped} skipped (already existed)')
```

Verify the count with aggregation:

```python
response = search_client.aggregate_data_objects(
    vs.AggregateDataObjectsRequest(
        parent=collection_parent,
        aggregate='COUNT',
    )
)
count = int(response.aggregate_results[0]['count'])
print(f'DataObjects in collection: {count}')
```

## Retrieval Without Index (Brute Force)

Without an ANN index, Vector Search 2 performs brute-force exact nearest neighbor (kNN) search across all DataObjects. This is slower for large collections but returns exact results — useful for development and establishing ground truth.

This section demonstrates all search and query methods available in VS2.

### Vector Search (BYOE)

Provide your own query vector to search against the `chunk_embedding` field. This is the traditional embedding-based retrieval approach — you compute the query embedding client-side and send the vector.

```python
pager = search_client.search_data_objects(
    vs.SearchDataObjectsRequest(
        parent=collection_parent,
        vector_search=vs.VectorSearch(
            search_field='chunk_embedding',
            vector={'values': question_embedding},
            top_k=5,
            output_fields=vs.OutputFields(
                data_fields=['chunk_id', 'gse', 'content'],
            ),
        ),
    )
)

results = []
for page in pager.pages:
    results.extend(page.results)

print(f'Question: {question}\n')
for i, r in enumerate(results):
    data = type(r.data_object).to_dict(r.data_object)['data']
    print(f'Match {i+1} (distance: {r.distance:.4f}):')
    print(f'  chunk_id: {data["chunk_id"]}')
    print(f'  gse: {data["gse"]}')
    print(f'  content: {data["content"][:150]}...')
    print()
```

#### Output Fields Control

The `output_fields` parameter controls which fields are returned. Use `*` to return all fields in each category:

```python
pager = search_client.search_data_objects(
    vs.SearchDataObjectsRequest(
        parent=collection_parent,
        vector_search=vs.VectorSearch(
            search_field='chunk_embedding',
            vector={'values': question_embedding},
            top_k=1,
            output_fields=vs.OutputFields(
                data_fields='*',
                vector_fields='*',
                metadata_fields='*',
            ),
        ),
    )
)

results = []
for page in pager.pages:
    results.extend(page.results)

result_dict = type(results[0].data_object).to_dict(results[0].data_object)
print('Data keys:', list(result_dict['data'].keys()))
print('Vector keys:', list(result_dict['vectors'].keys()))
print(f'Distance: {results[0].distance:.4f}')
print(f'Create time: {results[0].data_object.create_time}')
```

### Semantic Search (Auto-Embedding)

With semantic search, you provide natural language text and VS2 auto-embeds the query using the model configured in the collection's `vertex_embedding_config`. No client-side embedding call needed.

The `task_type` for queries should be `QUESTION_ANSWERING` (or `RETRIEVAL_QUERY`) — this is the asymmetric counterpart to the `RETRIEVAL_DOCUMENT` task type used at import time, and [can improve search quality by 30-40%](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/vector-search-2/embeddings/autogenerating-embeddings).

```python
pager = search_client.search_data_objects(
    vs.SearchDataObjectsRequest(
        parent=collection_parent,
        semantic_search=vs.SemanticSearch(
            search_text=question,
            search_field='auto_embedding',
            task_type='QUESTION_ANSWERING',
            top_k=5,
            output_fields=vs.OutputFields(
                data_fields=['chunk_id', 'gse', 'content'],
            ),
        ),
    )
)

results = []
for page in pager.pages:
    results.extend(page.results)

print(f'Question: {question}\n')
for i, r in enumerate(results):
    data = type(r.data_object).to_dict(r.data_object)['data']
    print(f'Match {i+1} (distance: {r.distance:.4f}):')
    print(f'  chunk_id: {data["chunk_id"]}')
    print(f'  gse: {data["gse"]}')
    print(f'  content: {data["content"][:150]}...')
    print()
```

### Text Search (Full-Text / Keyword)

### Pre-Filtered Search

Filters can be applied to any search type to scope results before similarity ranking. VS2 uses MongoDB-style filter operators:

| Operator | Description | Example |
|---|---|---|
| `$eq` | Equal to | `{"gse": {"$eq": "fannie"}}` |
| `$ne` | Not equal to | `{"gse": {"$ne": "freddie"}}` |
| `$gt`, `$gte` | Greater / greater-or-equal | `{"year": {"$gt": 2020}}` |
| `$lt`, `$lte` | Less / less-or-equal | `{"year": {"$lte": 2020}}` |
| `$in` | Value in list | `{"gse": {"$in": ["fannie", "freddie"]}}` |
| `$nin` | Value not in list | `{"gse": {"$nin": ["freddie"]}}` |
| `$and` | Logical AND | `{"$and": [{...}, {...}]}` |
| `$or` | Logical OR | `{"$or": [{...}, {...}]}` |
| `$all` | Array contains all values | `{"tags": {"$all": ["a", "b"]}}` |

## Hybrid Search

[Hybrid search](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/vector-search-2/query-search/search) uses `batch_search_data_objects()` to execute multiple searches in parallel and fuse results. This combines the strengths of different search types:
- **Semantic search** understands meaning — "lender obligations" finds content about "servicer responsibilities"
- **Text search** matches exact terms — finds specific phrases, names, or codes
- **Vector search** uses pre-computed embeddings for precise similarity

Results are merged using [Reciprocal Rank Fusion (RRF)](https://en.wikipedia.org/wiki/Reciprocal_rank_fusion), which combines ranked lists without needing tuned weights.

### Semantic + Text Search with RRF

```python
output = vs.OutputFields(data_fields=['chunk_id', 'gse', 'content'])

response = search_client.batch_search_data_objects(
    vs.BatchSearchDataObjectsRequest(
        parent=collection_parent,
        searches=[
            vs.Search(
                semantic_search=vs.SemanticSearch(
                    search_text=question,
                    search_field='auto_embedding',
                    task_type='QUESTION_ANSWERING',
                    top_k=10,
                    output_fields=output,
                ),
            ),
            vs.Search(
                text_search=vs.TextSearch(
                    search_text='lender servicing functions',
                    data_field_names=['content'],
                    top_k=10,
                    output_fields=output,
                ),
            ),
        ],
        combine=vs.BatchSearchDataObjectsRequest.CombineResultsOptions(
            ranker=vs.Ranker(
                rrf=vs.ReciprocalRankFusion(weights=[1.0, 1.0]),
            ),
        ),
    )
)

print(f'Hybrid Search: Semantic + Text with RRF\n')
for i, r in enumerate(response.results[0].results):
    data = type(r.data_object).to_dict(r.data_object)['data']
    print(f'Match {i+1} (distance: {r.distance:.4f}):')
    print(f'  chunk_id: {data["chunk_id"]} | gse: {data["gse"]}')
    print(f'  content: {data["content"][:150]}...')
    print()
```

### Semantic + Vector Search with RRF

Combine auto-embedding semantic search with BYOE vector search. This searches across both embedding fields:

```python
response = search_client.batch_search_data_objects(
    vs.BatchSearchDataObjectsRequest(
        parent=collection_parent,
        searches=[
            vs.Search(
                semantic_search=vs.SemanticSearch(
                    search_text=question,
                    search_field='auto_embedding',
                    task_type='QUESTION_ANSWERING',
                    top_k=10,
                    output_fields=output,
                ),
            ),
            vs.Search(
                vector_search=vs.VectorSearch(
                    search_field='chunk_embedding',
                    vector={'values': question_embedding},
                    top_k=10,
                    output_fields=output,
                ),
            ),
        ],
        combine=vs.BatchSearchDataObjectsRequest.CombineResultsOptions(
            ranker=vs.Ranker(
                rrf=vs.ReciprocalRankFusion(weights=[1.0, 1.0]),
            ),
        ),
    )
)

print(f'Hybrid Search: Semantic + Vector with RRF\n')
for i, r in enumerate(response.results[0].results):
    data = type(r.data_object).to_dict(r.data_object)['data']
    print(f'Match {i+1} (distance: {r.distance:.4f}):')
    print(f'  chunk_id: {data["chunk_id"]} | gse: {data["gse"]}')
    print(f'  content: {data["content"][:150]}...')
    print()
```

## Retrieval Augmented Generation (RAG)

Build a simple retrieval augmented generation process that enhances a query by retrieving context. This is done here by constructing three functions for the stages:
- `retrieve` - a function that uses semantic search to find matching context chunks
- `augment` - prepare chunks into a prompt
- `generate` - make the LLM request with the augmented prompt

A final function is used to combine the three stages and return the RAG response.

### Generation Model

Some models are only available in multi-region locations (e.g., `us`) rather than regional endpoints. We create a separate client for generation:

```python
GENERATION_MODEL = 'gemini-3.5-flash'

genai_client_us = genai.Client(vertexai=True, project=PROJECT_ID, location='us')
```

### Retrieve Function

```python
def retrieve(query, top_k=5):
    pager = search_client.search_data_objects(
        vs.SearchDataObjectsRequest(
            parent=collection_parent,
            semantic_search=vs.SemanticSearch(
                search_text=query,
                search_field='auto_embedding',
                task_type='QUESTION_ANSWERING',
                top_k=top_k,
                output_fields=vs.OutputFields(
                    data_fields=['chunk_id', 'content'],
                ),
            ),
        )
    )
    results = []
    for page in pager.pages:
        results.extend(page.results)
    return results
```

### Augment Function

```python
def augment(query, results):
    context = '\n\n---\n\n'.join([
        type(r.data_object).to_dict(r.data_object)['data']['content']
        for r in results
    ])
    prompt = f"""Answer the following question using only the context provided below.
If the answer is not in the context, say "I don't have enough information to answer that."

Context:
{context}

Question: {query}

Answer:"""
    return prompt
```

### Generate Function

```python
def generate(prompt):
    response = genai_client_us.models.generate_content(
        model=GENERATION_MODEL,
        contents=prompt,
    )
    return response.text
```

### RAG Function

```python
def rag(query, top_k=5):
    start = time.time()

    retrieve_start = time.time()
    results = retrieve(query, top_k)
    retrieve_time = time.time() - retrieve_start

    augment_start = time.time()
    prompt = augment(query, results)
    augment_time = time.time() - augment_start

    generate_start = time.time()
    answer = generate(prompt)
    generate_time = time.time() - generate_start

    total_time = time.time() - start

    return {
        'query': query,
        'answer': answer,
        'results': results,
        'timing': {
            'retrieve': retrieve_time,
            'augment': augment_time,
            'generate': generate_time,
            'total': total_time,
        },
    }
```

### Example In Use

```python
response = rag(question)
print(f'Question: {response["query"]}\n')
print(f'Answer:\n{response["answer"]}\n')
print(f'Timing:')
for step, t in response['timing'].items():
    print(f'  {step}: {t:.3f}s')
```

```python
response2 = rag("What are the requirements for flood insurance?")
print(f'Question: {response2["query"]}\n')
print(f'Answer:\n{response2["answer"]}\n')
print(f'Timing:')
for step, t in response2['timing'].items():
    print(f'  {step}: {t:.3f}s')
```

### Profiling Performance

Profile the timing of retrieval for sequential calls. The environment chosen for this workflow is serverless (no instances to manage), so we measure the service response time directly.

```python
n_tries = 100
retrieve_times = []

for i in range(n_tries):
    start = time.time()
    _ = retrieve(question)
    retrieve_times.append(time.time() - start)

retrieve_times = np.array(retrieve_times)
```
