# Retrieval - Vertex AI Vector Search 2 — RAG Engine

[RAG Engine](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/rag-engine/rag-overview) is a managed Retrieval Augmented Generation framework on Google Cloud that handles document parsing, chunking, embedding, storage, retrieval, and grounded generation. When configured with [Vector Search 2 as its backend](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/rag-engine/use-rag-managed-vertex-ai-vector-search), RAG Engine creates and manages a VS2 collection automatically — you provide documents and ask questions.

This workflow demonstrates:
- **RAG corpus creation** with `RagManagedVertexVectorSearch` as the vector backend
- **Document import** — RAG Engine parses, chunks, embeds, and indexes PDFs from GCS
- **Retrieval query** — retrieve relevant contexts from the corpus
- **Grounded generation** — Gemini generates answers using retrieved context as a tool
- **Inspect the managed VS2 collection** — use the VS2 SDK to see what RAG Engine created

The same [source documents](../../Chunking/readme.md) used across all retrieval workflows (Fannie Mae and Freddie Mac mortgage guidelines) are imported here, but RAG Engine handles chunking and embedding — no pre-processing needed.

> **Preview:** The `RagManagedVertexVectorSearch` backend is in Preview and currently available only in `us-central1`.

For the companion notebooks that work directly with the VS2 API, see [Retrieval - Vertex AI Vector Search 2](Retrieval%20-%20Vertex%20AI%20Vector%20Search%202.ipynb) and [Retrieval - Vertex AI Vector Search 2 — ANN Indexes](Retrieval%20-%20Vertex%20AI%20Vector%20Search%202%20-%20ANN%20Indexes.ipynb).

## Source Documents

RAG Engine handles document parsing, chunking, and embedding automatically. We point it to the source PDFs from the [chunking workflow](../../Chunking/readme.md) — 11 PDF parts of the Fannie Mae and Freddie Mac mortgage guidelines stored in GCS.

> **Prerequisite:** These PDFs were uploaded to GCS by the [Large Document Processing](../../Chunking/Large%20Document%20Processing%20-%20Document%20AI%20Layout%20Parser.ipynb) workflow. If they don't exist in your bucket, run that notebook first or substitute your own GCS document paths.

```python
from google.cloud import storage

gcs_client = storage.Client(project=PROJECT_ID)
bucket_name = PROJECT_ID
prefix = 'applied-genai/layout-parser-large-files/parts/'

blobs = list(gcs_client.list_blobs(bucket_name, prefix=prefix))
pdf_blobs = [b for b in blobs if b.name.endswith('.pdf')]

print(f'Found {len(pdf_blobs)} PDFs in gs://{bucket_name}/{prefix}')
for b in pdf_blobs:
    print(f'  {b.name} ({b.size / (1024*1024):.1f} MB)')

paths = [f'gs://{bucket_name}/{b.name}' for b in pdf_blobs]
```

## Configure RAG Engine Mode

RAG Engine in `us-central1` requires Serverless mode. Check the current configuration and switch if needed:

```python
config_name = f'projects/{PROJECT_ID}/locations/{REGION}/ragEngineConfig'
config = rag.get_rag_engine_config(name=config_name)
print(f'Current mode: {config.rag_managed_db_config.mode}')

if isinstance(config.rag_managed_db_config.mode, rag.Spanner):
    print('Switching to Serverless mode...')
    config.rag_managed_db_config = rag.RagManagedDbConfig(mode=rag.Serverless())
    updated = rag.update_rag_engine_config(rag_engine_config=config)
    print(f'Updated mode: {updated.rag_managed_db_config.mode}')
else:
    print('Already in Serverless mode')
```

## Create RAG Corpus with VS2 Backend

Create a RAG corpus backed by [Vector Search 2](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/rag-engine/use-rag-managed-vertex-ai-vector-search). RAG Engine creates and manages a VS2 collection automatically — you get full visibility into the collection in your project.

The embedding model is configured at corpus creation time. RAG Engine will use it to embed both imported documents and queries.

```python
embedding_model_config = rag.EmbeddingModelConfig(
    publisher_model=f'publishers/google/models/{EMBEDDING_MODEL}'
)

vector_db = rag.RagManagedVertexVectorSearch()

rag_corpus = rag.create_corpus(
    display_name=CORPUS_DISPLAY_NAME,
    backend_config=rag.RagVectorDbConfig(
        vector_db=vector_db,
        rag_embedding_model_config=embedding_model_config,
    ),
)

print(f'Corpus: {rag_corpus.name}')
print(f'Display name: {rag_corpus.display_name}')
```

---
## Import Documents

Import the source PDFs into the RAG corpus. RAG Engine handles:
1. **Parsing** — extracts text from PDFs
2. **Chunking** — splits text into chunks (configurable size and overlap)
3. **Embedding** — generates embeddings using the configured model
4. **Indexing** — stores chunks and embeddings in the managed VS2 collection

```python
response = rag.import_files(
    corpus_name=rag_corpus.name,
    paths=paths,
    transformation_config=rag.TransformationConfig(
        chunking_config=rag.ChunkingConfig(
            chunk_size=512,
            chunk_overlap=100,
        ),
    ),
    max_embedding_requests_per_min=1000,
)

print(f'Import complete: {response}')
```

Verify imported files:

```python
files = list(rag.list_files(corpus_name=rag_corpus.name))

print(f'Imported {len(files)} files:\n')
for f in files:
    print(f'  {f.display_name}')
    print(f'    {f.name}')
```

---
## Retrieval Query

Retrieve relevant contexts directly from the corpus using `rag.retrieval_query()`. This embeds the query, searches the managed VS2 collection, and returns matching chunks with their source information.

```python
question = "Does a lender have to perform servicing functions directly?"

rag_retrieval_config = rag.RagRetrievalConfig(
    top_k=5,
    filter=rag.Filter(vector_distance_threshold=0.5),
)

response = rag.retrieval_query(
    rag_resources=[
        rag.RagResource(rag_corpus=rag_corpus.name),
    ],
    text=question,
    rag_retrieval_config=rag_retrieval_config,
)

print(f'Question: {question}\n')
for i, context in enumerate(response.contexts.contexts):
    print(f'Context {i+1} (distance: {context.distance:.4f}):')
    print(f'  Source: {context.source_display_name}')
    print(f'  Text: {context.text[:200]}...')
    print()
```

---
## Generation with Retrieved Context

Use the contexts retrieved by `rag.retrieval_query()` to ground Gemini's responses. The generation model (`gemini-3.5-flash`) is only available in multi-region locations, so we use a separate `genai` client pointed at `us`.

```python
def retrieve_and_generate(query, top_k=10):
    contexts = rag.retrieval_query(
        rag_resources=[rag.RagResource(rag_corpus=rag_corpus.name)],
        text=query,
        rag_retrieval_config=rag.RagRetrievalConfig(
            top_k=top_k,
            filter=rag.Filter(vector_distance_threshold=0.5),
        ),
    )

    context_text = '\n\n---\n\n'.join([
        c.text for c in contexts.contexts.contexts
    ])

    prompt = f"""Use the following context to answer the question. If the context doesn't contain enough information, say so.

Context:
{context_text}

Question: {query}"""

    response = genai_client_us.models.generate_content(
        model=GENERATION_MODEL,
        contents=prompt,
    )
    return response.text
```

```python
answer = retrieve_and_generate(question)
print(f'Question: {question}\n')
print(f'Answer:\n{answer}')
```

```python
question2 = "What are the requirements for flood insurance?"

answer2 = retrieve_and_generate(question2)
print(f'Question: {question2}\n')
print(f'Answer:\n{answer2}')
```

```python
question3 = "How is the loan-to-value ratio calculated?"

answer3 = retrieve_and_generate(question3)
print(f'Question: {question3}\n')
print(f'Answer:\n{answer3}')
```

## Inspect the Managed VS2 Collection

RAG Engine created a VS2 collection in your project. You can inspect it directly using the VS2 SDK — the same API used in the [main VS2 notebook](Retrieval%20-%20Vertex%20AI%20Vector%20Search%202.ipynb). This gives you full visibility into the managed infrastructure.

```python
vs_client = vs.VectorSearchServiceClient()
search_client = vs.DataObjectSearchServiceClient()

parent = f'projects/{PROJECT_ID}/locations/{REGION}'

print('VS2 Collections in project:\n')
for collection in vs_client.list_collections(parent=parent):
    col_id = collection.name.split('/collections/')[-1]
    print(f'  {col_id}')
    print(f'    Name: {collection.name}')
    print(f'    Data schema: {list(collection.data_schema.get("properties", {}).keys())}')
    print(f'    Vector schema: {list(collection.vector_schema.get("properties", {}).keys())}')
    print()
```

```python
for collection in vs_client.list_collections(parent=parent):
    col_name = collection.name
    col_id = col_name.split('/collections/')[-1]

    try:
        response = search_client.aggregate_data_objects(
            vs.AggregateDataObjectsRequest(
                parent=col_name,
                aggregate='COUNT',
            )
        )
        print(f'Collection {col_id}: {response.count} DataObjects')
    except Exception as e:
        print(f'Collection {col_id}: {e}')
```
