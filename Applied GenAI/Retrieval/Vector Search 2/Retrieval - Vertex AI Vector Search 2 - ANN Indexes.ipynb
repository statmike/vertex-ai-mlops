# Retrieval - Vertex AI Vector Search 2 — ANN Indexes

This notebook is a companion to [Retrieval - Vertex AI Vector Search 2](Retrieval%20-%20Vertex%20AI%20Vector%20Search%202.ipynb), which creates a VS2 **Collection** and loads 9,040 DataObjects with BYOE and auto-embedded vectors. **Run that notebook first** — this one assumes the collection already exists.

Here we focus on [ANN (Approximate Nearest Neighbor) indexes](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/vector-search-2/indexes/indexes), which accelerate search at scale using Google's ScaNN algorithm. Without an index, VS2 performs brute-force kNN — exact but slower for large collections. ANN trades a tiny amount of accuracy for massive speed improvements.

This workflow demonstrates:
- **ANN index creation** on both BYOE and auto-embedding vector fields
- **Vector search** — BYOE query vector against ANN index
- **Semantic search** — auto-embedded text query against ANN index
- **Pre-filtered search** — scoped ANN search with MongoDB-style filters
- **Recall calibration** — measure recall@K by comparing ANN results to brute-force kNN ground truth

## Save Ground Truth (kNN) Results

Before creating ANN indexes, capture brute-force kNN results as ground truth. After indexes are created, we compare ANN results against these to measure **recall** — the fraction of true nearest neighbors that the approximate index finds.

```python
recall_queries = [
    "Does a lender have to perform servicing functions directly?",
    "What are the requirements for flood insurance?",
    "How is the loan-to-value ratio calculated?",
    "What are the guidelines for property appraisal?",
    "When can a borrower request cancellation of mortgage insurance?",
]

RECALL_K = 20

def get_knn_ids(query_text, search_field='auto_embedding', top_k=RECALL_K):
    pager = search_client.search_data_objects(
        vs.SearchDataObjectsRequest(
            parent=collection_parent,
            semantic_search=vs.SemanticSearch(
                search_text=query_text,
                search_field=search_field,
                task_type='QUESTION_ANSWERING',
                top_k=top_k,
                output_fields=vs.OutputFields(data_fields=['chunk_id']),
            ),
        )
    )
    results = []
    for page in pager.pages:
        results.extend(page.results)
    return [type(r.data_object).to_dict(r.data_object)['data']['chunk_id'] for r in results]

ground_truth = {}
for q in recall_queries:
    ground_truth[q] = get_knn_ids(q)
    print(f'Saved {len(ground_truth[q])} kNN results for: "{q[:60]}..."')

print(f'\nGround truth saved for {len(ground_truth)} queries at top_k={RECALL_K}')
```

*The ground truth results above were captured with brute-force kNN (no index). After creating ANN indexes below, we will compare ANN results against these to measure recall.*

## Create ANN Index

An [ANN (Approximate Nearest Neighbor) index](https://docs.cloud.google.com/gemini-enterprise-agent-platform/build/vector-search-2/indexes/indexes) accelerates search at scale using Google's ScaNN algorithm. Without an index, search uses brute-force kNN. With an index, VS2 uses ANN — trading a tiny amount of accuracy (~99%) for massive speed gains.

Index configuration:
- **`index_field`** — which vector field to index
- **`filter_fields`** — fields available for inline pre-filtering during search (default: all string/numeric/boolean fields)
- **`store_fields`** — payload fields returned with results without a separate lookup
- **Distance metric** — defaults to `DOT_PRODUCT` (use the metric your embedding model was trained on)

We create two indexes — one for each vector field:

```python
index_configs = [
    {
        'index_id': 'chunk-embedding-index',
        'index_field': 'chunk_embedding',
        'filter_fields': ['gse'],
        'store_fields': ['chunk_id', 'content', 'gse'],
        'description': 'BYOE embedding index',
    },
    {
        'index_id': 'auto-embedding-index',
        'index_field': 'auto_embedding',
        'filter_fields': ['gse'],
        'store_fields': ['chunk_id', 'content', 'gse'],
        'description': 'Auto-embedding index',
    },
]

existing_indexes = {
    idx.name.split('/')[-1]: idx
    for idx in vs_client.list_indexes(parent=collection_parent)
}

for config in index_configs:
    index_id = config['index_id']
    print(f'\nIndex: {index_id} ({config["description"]})')

    if index_id in existing_indexes:
        print(f'  Already exists — skipping')
        continue

    try:
        operation = vs_client.create_index(
            parent=collection_parent,
            index_id=index_id,
            index=vs.Index(
                index_field=config['index_field'],
                filter_fields=config['filter_fields'],
                store_fields=config['store_fields'],
            ),
        )
    except Exception as e:
        print(f'  create_index raised {type(e).__name__}: {e}')
        print(f'  Checking for in-progress operation...')
        ops = vs_client.list_operations(
            request={'name': collection_parent + '/operations', 'filter': f'done=false'}
        )
        matching_op = None
        for op in ops.operations:
            if index_id in op.name:
                matching_op = op
                break
        if matching_op is None:
            for op in ops.operations:
                matching_op = matching_op or op
        if matching_op:
            from google.api_core import operation as ga_operation
            operation = ga_operation.from_gapic(
                matching_op,
                vs_client._transport.operations_client,
                vs.Index,
            )
            print(f'  Found operation: {matching_op.name}')
        else:
            print(f'  No in-progress operation found — skipping')
            continue

    print(f'  Waiting for index build (up to 90 min)...')
    try:
        result = operation.result(timeout=5400)
        print(f'  Done: {result.name}')
    except TimeoutError:
        print(f'  Timed out — index is still building. Re-run this cell to check again.')
```

Verify indexes:

```python
for idx in vs_client.list_indexes(parent=collection_parent):
    idx_id = idx.name.split('/indexes/')[-1]
    print(f'Index: {idx_id}')
    print(f'  Field: {idx.index_field}')
    print(f'  Distance: {idx.distance_metric.name}')
    print(f'  Filter fields: {list(idx.filter_fields)}')
    print(f'  Store fields: {list(idx.store_fields)}')
    print()
```

## Retrieval With Index (ANN)

Now that ANN indexes are active, the same search operations will use approximate nearest neighbor search instead of brute force. This should be faster, especially for larger collections, with nearly identical results.

### Vector Search (ANN)

```python
pager = search_client.search_data_objects(
    vs.SearchDataObjectsRequest(
        parent=collection_parent,
        vector_search=vs.VectorSearch(
            search_field='chunk_embedding',
            vector={'values': question_embedding},
            top_k=5,
            output_fields=vs.OutputFields(
                data_fields=['chunk_id', 'gse', 'content'],
            ),
        ),
    )
)

results = []
for page in pager.pages:
    results.extend(page.results)

print(f'Vector Search (ANN) — Question: {question}\n')
for i, r in enumerate(results):
    data = type(r.data_object).to_dict(r.data_object)['data']
    print(f'Match {i+1} (distance: {r.distance:.4f}):')
    print(f'  chunk_id: {data["chunk_id"]} | gse: {data["gse"]}')
    print(f'  content: {data["content"][:150]}...')
    print()
```

### Semantic Search (ANN)

```python
pager = search_client.search_data_objects(
    vs.SearchDataObjectsRequest(
        parent=collection_parent,
        semantic_search=vs.SemanticSearch(
            search_text=question,
            search_field='auto_embedding',
            task_type='QUESTION_ANSWERING',
            top_k=5,
            output_fields=vs.OutputFields(
                data_fields=['chunk_id', 'gse', 'content'],
            ),
        ),
    )
)

results = []
for page in pager.pages:
    results.extend(page.results)

print(f'Semantic Search (ANN) — Question: {question}\n')
for i, r in enumerate(results):
    data = type(r.data_object).to_dict(r.data_object)['data']
    print(f'Match {i+1} (distance: {r.distance:.4f}):')
    print(f'  chunk_id: {data["chunk_id"]} | gse: {data["gse"]}')
    print(f'  content: {data["content"][:150]}...')
    print()
```

### Pre-Filtered Search (ANN)

```python
pager = search_client.search_data_objects(
    vs.SearchDataObjectsRequest(
        parent=collection_parent,
        semantic_search=vs.SemanticSearch(
            search_text=question,
            search_field='auto_embedding',
            task_type='QUESTION_ANSWERING',
            top_k=5,
            filter={'gse': {'$eq': 'fannie'}},
            output_fields=vs.OutputFields(
                data_fields=['chunk_id', 'gse', 'content'],
            ),
        ),
    )
)

results = []
for page in pager.pages:
    results.extend(page.results)

print(f'Semantic Search (ANN) filtered to gse="fannie":\n')
for i, r in enumerate(results):
    data = type(r.data_object).to_dict(r.data_object)['data']
    print(f'Match {i+1} (distance: {r.distance:.4f}):')
    print(f'  chunk_id: {data["chunk_id"]} | gse: {data["gse"]}')
    print(f'  content: {data["content"][:150]}...')
    print()
```

### Recall Calibration: kNN vs ANN

With both brute-force kNN (ground truth) and ANN index results available on the same collection, we can measure **recall@K** — the fraction of the true top-K nearest neighbors that the ANN index returns. This is the standard metric for evaluating approximate nearest neighbor quality.

$$\text{recall@K} = \frac{|\text{ANN top-K} \cap \text{kNN top-K}|}{K}$$

A recall of 1.0 means the ANN index returned exactly the same results as brute force. VS2's ScaNN-based indexes typically achieve 95-99%+ recall.

```python
def get_ann_ids(query_text, search_field='auto_embedding', top_k=RECALL_K):
    pager = search_client.search_data_objects(
        vs.SearchDataObjectsRequest(
            parent=collection_parent,
            semantic_search=vs.SemanticSearch(
                search_text=query_text,
                search_field=search_field,
                task_type='QUESTION_ANSWERING',
                top_k=top_k,
                output_fields=vs.OutputFields(data_fields=['chunk_id']),
            ),
        )
    )
    results = []
    for page in pager.pages:
        results.extend(page.results)
    return [type(r.data_object).to_dict(r.data_object)['data']['chunk_id'] for r in results]

recall_results = []
for q in recall_queries:
    ann_ids = get_ann_ids(q)
    knn_ids = ground_truth[q]
    overlap = len(set(ann_ids) & set(knn_ids))
    recall = overlap / len(knn_ids)
    recall_results.append({
        'query': q[:65] + '...' if len(q) > 65 else q,
        'knn_count': len(knn_ids),
        'ann_count': len(ann_ids),
        'overlap': overlap,
        f'recall@{RECALL_K}': recall,
    })

recall_df = pd.DataFrame(recall_results)
avg_recall = recall_df[f'recall@{RECALL_K}'].mean()

print(f'Recall Calibration: kNN (brute force) vs ANN (index) at top_k={RECALL_K}\n')
print(recall_df.to_string(index=False))
print(f'\nAverage recall@{RECALL_K}: {avg_recall:.4f}')
```
