Features computed in BigQuery are only useful for real-time serving if they reach Bigtable fast enough. This notebook explores the spectrum of sync strategies — from one-time backfill to sub-minute continuous updates — and measures the freshness tradeoffs.

## Section 1: One-Time Export (EXPORT DATA)

The simplest synchronization pattern is a one-time snapshot: compute all features in BigQuery, export them to Bigtable, done. This is ideal for:

- **Initial backfill** — populating the feature store for the first time
- **Full refresh** — replacing all features after a model retrain or schema change
- **Ad-hoc updates** — manually refreshing features when data quality issues are fixed

The mechanism is BigQuery's [EXPORT DATA](https://cloud.google.com/bigquery/docs/reference/standard-sql/other-statements#export_data_statement) statement with `format = 'CLOUD_BIGTABLE'`. This writes query results directly to Bigtable without any intermediate storage.

### EXPORT DATA Options

| Option | Value | Purpose |
|--------|-------|--------|
| `format` | `'CLOUD_BIGTABLE'` | Target format — writes to Bigtable instead of GCS |
| `uri` | Bigtable URI with app profile | Target table and routing config |
| `overwrite` | `true` or `false` | `true` = delete existing rows first (full replace); `false` = upsert (insert or update matching row keys) |

The URI includes the app profile, which controls cluster routing and request priority:

```
https://bigtable.googleapis.com/projects/{PROJECT}/instances/{INSTANCE}/appProfiles/{PROFILE}/tables/{TABLE}
```

### Infrastructure Requirements

EXPORT DATA to Bigtable requires two pieces of infrastructure:

1. **Enterprise reservation** — BigQuery Enterprise edition is required. We create a reservation with slots and a QUERY-type assignment. Slots are billed per-second at on-demand Enterprise rates.

2. **Bigtable app profile** — configured with single-cluster routing and LOW request priority so export traffic does not compete with latency-sensitive serving reads.

### Create Enterprise Reservation

No capacity commitment needed — on-demand edition billing. Slots are billed per-second while the reservation exists.

````python
# Create Enterprise reservation for EXPORT DATA
parent = f'projects/{PROJECT_ID}/locations/{LOCATION}'

try:
    reservation = reservation_client.create_reservation(
        parent=parent,
        reservation_id=RESERVATION_ID,
        reservation=bigquery_reservation_v1.Reservation(
            slot_capacity=RESERVATION_SLOTS,
            edition=bigquery_reservation_v1.Edition.ENTERPRISE,
        ),
    )
    print(f'Created reservation: {reservation.name}')
    print(f'  Slots:   {reservation.slot_capacity}')
    print(f'  Edition: ENTERPRISE')
except AlreadyExists:
    reservation = reservation_client.get_reservation(
        name=f'{parent}/reservations/{RESERVATION_ID}'
    )
    print(f'Reservation exists: {reservation.name}')
````

````python
# QUERY assignment — routes query jobs through the reservation
try:
    assignment = reservation_client.create_assignment(
        parent=reservation.name,
        assignment=bigquery_reservation_v1.Assignment(
            job_type=bigquery_reservation_v1.Assignment.JobType.QUERY,
            assignee=f'projects/{PROJECT_ID}',
        ),
    )
    print(f'Created assignment: {assignment.name}')
    print(f'  Job type: QUERY')
    print(f'  Assignee: {PROJECT_ID}')
except AlreadyExists:
    assignments = list(reservation_client.list_assignments(parent=reservation.name))
    assignment = assignments[0]
    print(f'Assignment exists: {assignment.name}')

print(f'\nReservation ready:')
print(f'  Reservation: {RESERVATION_ID} ({RESERVATION_SLOTS} slots, Enterprise)')
print(f'  Assignment:  QUERY \u2192 {PROJECT_ID}')

# Reservation assignments take a moment to propagate
print('\nWaiting 60s for assignment to propagate...')
time.sleep(60)
print('Ready.')
````

### Create Bigtable App Profile

The [app profile](https://cloud.google.com/bigtable/docs/app-profiles) controls how requests are routed to clusters. EXPORT DATA requires:
- **Single-cluster routing** — writes go to exactly one cluster (no multi-cluster replication conflicts)
- **LOW request priority** — export traffic yields to serving reads

We use `gcloud` to create the profile since the Python client's app profile API requires more boilerplate.

````python
# Create app profile for EXPORT DATA
import subprocess

result = subprocess.run([
    'gcloud', 'bigtable', 'app-profiles', 'create', APP_PROFILE_ID,
    '--instance', BT_INSTANCE,
    '--project', PROJECT_ID,
    '--route-to', f'{BT_INSTANCE}-c1',
    '--priority', 'PRIORITY_LOW',
    '--standard',
    '--description', 'BigQuery EXPORT DATA - single cluster, LOW priority',
    '--force',
], capture_output=True, text=True)

if result.returncode == 0:
    print(f'Created app profile: {APP_PROFILE_ID}')
elif 'already exists' in result.stderr.lower():
    print(f'App profile exists: {APP_PROFILE_ID}')
else:
    print(f'Error: {result.stderr.strip()}')
    # Try without --standard flag (older gcloud versions)
    result2 = subprocess.run([
        'gcloud', 'bigtable', 'app-profiles', 'create', APP_PROFILE_ID,
        '--instance', BT_INSTANCE,
        '--project', PROJECT_ID,
        '--route-to', f'{BT_INSTANCE}-c1',
        '--description', 'BigQuery EXPORT DATA - single cluster routing',
        '--force',
    ], capture_output=True, text=True)
    if result2.returncode == 0:
        print(f'Created app profile: {APP_PROFILE_ID} (without priority flag)')
    elif 'already exists' in result2.stderr.lower():
        print(f'App profile exists: {APP_PROFILE_ID}')
    else:
        print(f'Error: {result2.stderr.strip()}')

# Verify
result = subprocess.run([
    'gcloud', 'bigtable', 'app-profiles', 'describe', APP_PROFILE_ID,
    '--instance', BT_INSTANCE,
    '--project', PROJECT_ID,
], capture_output=True, text=True)
if result.returncode == 0:
    print(f'\nProfile details:')
    for line in result.stdout.strip().split('\n'):
        print(f'  {line}')
````

### Build and Execute the Export

The export query selects a subset of features from the `dense_features` table (created in Notebook 1) and writes them to Bigtable. The query uses the STRUCT pattern to map columns to Bigtable's column family and qualifiers:

- The first column (`rowkey`) becomes the **Bigtable row key**
- The STRUCT name (`features`) becomes the **column family**
- Each STRUCT field becomes a **column qualifier**

````python
# Bigtable URI for EXPORT DATA (includes app profile)
BIGTABLE_URI = (
    f'https://bigtable.googleapis.com/projects/{PROJECT_ID}'
    f'/instances/{BT_INSTANCE}'
    f'/appProfiles/{APP_PROFILE_ID}'
    f'/tables/{BT_TABLE}'
)

print(f'Target URI: {BIGTABLE_URI}')
````

````python
# Build the EXPORT DATA query
# Export a subset of dense_features to Bigtable
EXPORT_SQL = f'''
EXPORT DATA OPTIONS (
    format = 'CLOUD_BIGTABLE',
    uri = '{BIGTABLE_URI}',
    overwrite = true
) AS
SELECT
    CAST(entity_id AS STRING) AS rowkey,
    STRUCT(
        CAST(feature_timestamp AS STRING) AS _ts,
        CONCAT(
            CAST(feature_1 AS STRING), '|',
            CAST(feature_2 AS STRING), '|',
            CAST(feature_3 AS STRING), '|',
            CAST(feature_4 AS STRING), '|',
            CAST(feature_5 AS STRING)
        ) AS _values
    ) AS {BT_COLUMN_FAMILY}
FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
'''

print('EXPORT DATA query:')
print(EXPORT_SQL)
````

````python
# Execute the export
print('Exporting features to Bigtable...')
print(f'  Source: {BQ_PROJECT}.{BQ_DATASET}.dense_features')
print(f'  Target: {BT_INSTANCE}/{BT_TABLE}')
print(f'  Mode:   overwrite = true (full replace)')
print()

start = time.perf_counter()
job = bq.query(EXPORT_SQL)
job.result()  # Wait for completion
export_time = time.perf_counter() - start

print(f'Export complete in {export_time:.1f}s')
print(f'  Job ID:          {job.job_id}')
print(f'  Bytes processed: {job.total_bytes_processed:,}')
if job.slot_millis:
    print(f'  Slot time:       {job.slot_millis:,} ms')
    print(f'  Slot seconds:    {job.slot_millis / 1000:.1f}s')
````

````python
# Verify: read a sample of rows from Bigtable
partial_rows = table.read_rows(
    filter_=CellsColumnLimitFilter(1),
    limit=10,
)
partial_rows.consume_all()

print(f'Sample rows from Bigtable ({len(partial_rows.rows)} rows):')
for key, row in list(partial_rows.rows.items())[:5]:
    cells = row.cells.get(BT_COLUMN_FAMILY, {})
    values = {}
    for col_name, col_cells in cells.items():
        values[col_name.decode('utf-8')] = col_cells[0].value.decode('utf-8')
    print(f'  {key.decode("utf-8"):>10s}: {values}')
````

### Post-Export Validation

After any export, verify data integrity: row count matches, values are correct, and the schema is consistent. This is especially important for scheduled exports where silent failures can go unnoticed.

````python
# Post-export validation: compare BigQuery source to Bigtable target
from google.cloud.bigtable.row_set import RowSet, RowRange
from google.cloud.bigtable.row_filters import CellsColumnLimitFilter
import random

print('=== Post-Export Validation ===\n')

# Check 1: Row count comparison
bq_count = bq.query(f'''
    SELECT COUNT(*) as cnt FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
''').to_dataframe()['cnt'][0]

bt_rows = list(table.read_rows(filter_=CellsColumnLimitFilter(1)))
bt_data_rows = [r for r in bt_rows if not r.row_key.decode().startswith('#')]
bt_count = len(bt_data_rows)

count_match = bt_count >= bq_count * 0.99  # Allow 1% tolerance for timing
print(f'1. Row count:  BQ={bq_count:,}  BT={bt_count:,}  {"PASS" if count_match else "WARN — counts differ"}')

# Check 2: Sample verification (10 random entities)
sample_df = bq.query(f'''
    SELECT entity_group, entity_id, feature_float_1
    FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
    ORDER BY RAND()
    LIMIT 10
''').to_dataframe()

sample_match = 0
for _, row_data in sample_df.iterrows():
    key = f'{row_data["entity_group"]}#{row_data["entity_id"]}'
    bt_row = table.read_row(key.encode(), filter_=CellsColumnLimitFilter(1))
    if bt_row and BT_COLUMN_FAMILY.encode() in bt_row.cells:
        sample_match += 1

print(f'2. Sample check: {sample_match}/{len(sample_df)} entities found in Bigtable  {"PASS" if sample_match == len(sample_df) else "WARN"}')

# Check 3: Schema consistency
schema_row = table.read_row('#schema'.encode(), filter_=CellsColumnLimitFilter(1))
if schema_row:
    schema_json = json.loads(schema_row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode())
    print(f'3. Schema row:  {len(schema_json.get("columns", []))} columns defined  PASS')
else:
    print(f'3. Schema row:  NOT FOUND — run the schema write cell first')

print(f'\n{"All checks passed!" if count_match and sample_match == len(sample_df) else "Review warnings above."}')
````

### Full Replace vs Incremental Upsert

The `overwrite` option in [`EXPORT DATA`](https://cloud.google.com/bigquery/docs/export-to-bigtable) controls how exported rows interact with existing data in Bigtable:

| | `overwrite = true` (Full Replace) | `overwrite = false` (Incremental Upsert) |
|---|---|---|
| **Behavior** | Deletes all existing rows in the table, then writes the exported rows | Writes exported rows on top of existing data — matching row keys are updated, non-matching rows are left untouched |
| **Data during export** | Brief window where Bigtable has **zero rows** (between delete and re-insert) | Existing data remains available throughout — no downtime |
| **Use case** | Initial backfill, schema migration, full refresh after model retrain | Scheduled syncs, incremental updates, change-based exports |
| **Risk** | Serving applications see empty results during the delete/re-insert window | Stale rows that no longer exist in BigQuery remain in Bigtable (no automatic deletion) |
| **Cleanup** | Not needed — old data is deleted automatically | May need periodic full replace or explicit row deletion to remove stale entities |

**Which to use:**

```
Is this the first time populating the table?
│
├─ Yes ──→ overwrite = true  (clean start)
│
└─ No
    │
    ├─ Can the serving app tolerate a brief gap? ──→ overwrite = true  (simplest)
    │
    └─ Need zero-downtime updates? ──→ overwrite = false  (upsert mode)
```

> **Continuous queries** (Section 3) always operate in upsert mode — they re-compute and re-export features only for entities with new activity, leaving all other rows untouched. This is the most efficient pattern for keeping features fresh because it only writes what changed.

### One-Time Export: Summary

The one-time export is the simplest synchronization pattern. It works well for:

| Use Case | Why it Works |
|----------|-------------|
| **Initial backfill** | Populate the entire feature store from historical data |
| **Full refresh after retrain** | Replace all features with newly computed values |
| **Schema migration** | Export with new column layout after schema changes |

The limitation is obvious: features are frozen at export time. Any new data that arrives after the export is invisible to serving. For features that change slowly (user demographics, lifetime aggregates), this may be acceptable. For features that change quickly (velocity counts, recent spending), we need something better.

---
## Section 2: Scheduled Sync

The next step up from one-time export is **scheduled sync**: run the same EXPORT DATA statement on a recurring schedule. This gives predictable freshness without the complexity of continuous queries.

BigQuery provides two mechanisms for this:

1. **Scheduled Queries** (via [BigQuery Data Transfer Service](https://cloud.google.com/bigquery/docs/scheduling-queries)) — native, no additional infrastructure
2. **Cloud Scheduler + Cloud Functions** — more flexible, supports custom logic before/after the export

### Freshness vs. Complexity Tradeoff

| Schedule Interval | Feature Freshness | Cost Implication |
|-------------------|-------------------|------------------|
| Every 24 hours | Features up to 1 day stale | Lowest — one job per day |
| Every 1 hour | Features up to 1 hour stale | Moderate — 24 jobs per day |
| Every 15 minutes | Features up to 15 min stale | Higher — 96 jobs per day |
| Every 5 minutes | Features up to 5 min stale | Highest for scheduled — 288 jobs per day |

Each job incurs the reservation slot cost for its duration. For small feature tables, each export may only take 10-30 seconds, so even frequent schedules are inexpensive.

### Scheduled Query via Python API

The [BigQuery Data Transfer Service](https://cloud.google.com/bigquery/docs/scheduling-queries) supports scheduling any SQL statement, including EXPORT DATA. Here is the API pattern for creating a scheduled export.

````python
# Scheduled query pattern using BigQuery Data Transfer Service
# This cell shows the API call pattern but does NOT create a live schedule.

from google.cloud import bigquery_datatransfer_v1
from google.protobuf.struct_pb2 import Struct

# The EXPORT DATA SQL to run on schedule
SCHEDULED_EXPORT_SQL = f'''
EXPORT DATA OPTIONS (
    format = 'CLOUD_BIGTABLE',
    uri = '{BIGTABLE_URI}',
    overwrite = false
) AS
SELECT
    CAST(entity_id AS STRING) AS rowkey,
    STRUCT(
        CAST(feature_timestamp AS STRING) AS _ts,
        CONCAT(
            CAST(feature_1 AS STRING), '|',
            CAST(feature_2 AS STRING), '|',
            CAST(feature_3 AS STRING), '|',
            CAST(feature_4 AS STRING), '|',
            CAST(feature_5 AS STRING)
        ) AS _values
    ) AS {BT_COLUMN_FAMILY}
FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
'''

# Build the transfer config (NOT executed — for reference)
transfer_config_params = Struct()
transfer_config_params.update({
    'query': SCHEDULED_EXPORT_SQL,
})

transfer_config = bigquery_datatransfer_v1.TransferConfig(
    display_name='Feature Store Sync — Every Hour',
    data_source_id='scheduled_query',
    schedule='every 1 hours',  # Options: 'every 15 minutes', 'every 6 hours', 'every day 02:00'
    params=transfer_config_params,
    destination_dataset_id=BQ_DATASET,  # Required by API even for EXPORT DATA
)

print('Scheduled query configuration (reference only):')
print(f'  Display name: {transfer_config.display_name}')
print(f'  Schedule:     {transfer_config.schedule}')
print(f'  Data source:  {transfer_config.data_source_id}')
print(f'  overwrite:    false (upsert mode — only updates changed rows)')
print()
print('To create this schedule, uncomment and run:')
print('  transfer_client = bigquery_datatransfer_v1.DataTransferServiceClient()')
print(f'  transfer_client.create_transfer_config(')
print(f'      parent="projects/{PROJECT_ID}/locations/{LOCATION}",')
print(f'      transfer_config=transfer_config,')
print(f'  )')
````

### Key Design Decisions for Scheduled Sync

**`overwrite = false` (upsert mode):** For scheduled syncs, upsert is usually better than full replace. It only updates rows whose features have changed, reducing write volume and avoiding a brief window where Bigtable has zero rows (between the delete-all and re-insert of `overwrite = true`).

**Schedule interval:** Match the freshness requirement of your features. Slowly-changing features (user demographics, lifetime aggregates) can sync daily. Velocity features (transaction counts, spending rates) need more frequent updates — but if you need sub-minute freshness, skip ahead to continuous queries.

### Alternative: Cloud Scheduler + Cloud Functions

For more control over the sync process, you can use [Cloud Scheduler](https://cloud.google.com/scheduler/docs/overview) to trigger a [Cloud Function](https://cloud.google.com/functions/docs/concepts/overview) that runs the EXPORT DATA query:

```
Cloud Scheduler (cron) → Cloud Functions (Python) → BigQuery EXPORT DATA → Bigtable
```

This adds flexibility:
- **Pre-export validation** — check data quality before exporting
- **Post-export notifications** — send Slack/email alerts on completion or failure
- **Conditional exports** — only run if upstream data has actually changed
- **Multi-step pipelines** — compute features in one step, export in another

The tradeoff is more infrastructure to manage. For most cases, the native Scheduled Query is sufficient.

### Orchestration Alternatives

While BigQuery's built-in scheduled queries are the simplest approach, production teams often need more control over dependencies, retries, and multi-step pipelines:

| Orchestrator | When to Use | Documentation |
|-------------|------------|---------------|
| **BQ Scheduled Queries** | Simple single-query schedules, no external dependencies | [Scheduled queries](https://cloud.google.com/bigquery/docs/scheduling-queries) |
| **Cloud Composer (Airflow)** | Multi-step DAGs: validate data → export → verify → notify. Handles dependencies between BQ, Bigtable, and other systems. | [Cloud Composer](https://cloud.google.com/composer/docs/concepts/overview) |
| **Cloud Scheduler + Cloud Functions** | Lightweight: trigger an EXPORT DATA job on a cron schedule without a full orchestration layer | [Cloud Scheduler](https://cloud.google.com/scheduler/docs/overview) |
| **Cloud Workflows** | Serverless orchestration for simple multi-step pipelines without managing Airflow infrastructure | [Workflows](https://cloud.google.com/workflows/docs/overview) |

> **Airflow pattern:** A typical Composer DAG would: (1) run a BQ query to identify changed entities, (2) run `EXPORT DATA` for those entities, (3) validate row counts in Bigtable, (4) send a Pub/Sub notification or Slack alert on completion/failure.

## Section 3: Continuous Queries

> **Reference pattern** — the code in this section shows the continuous query SQL and reservation configuration but does **not** execute the continuous query. The CONTINUOUS reservation type may not be available in all environments and runs indefinitely (consuming slots continuously).

BigQuery [continuous queries](https://cloud.google.com/bigquery/docs/continuous-queries-introduction) bridge the gap between scheduled batch exports and full streaming pipelines. A continuous query watches for new rows via the `APPENDS()` function and re-exports affected features within seconds of data arrival.

### How It Works

| Component | One-Time Export | Continuous Query |
|-----------|----------------|------------------|
| **Source** | Full table scan | `APPENDS(TABLE, timestamp)` — only new rows |
| **Reservation type** | QUERY assignment | CONTINUOUS assignment |
| **Lifecycle** | Runs once, completes | Runs indefinitely |
| **Freshness** | Snapshot at export time | Seconds after new rows arrive |
| **Cost** | Pay for one job | Pay for continuous slot allocation |

### Reservation Type Change

Continuous queries require a CONTINUOUS assignment (not QUERY). The CONTINUOUS assignment type has different constraints:
- Maximum 500 slots per reservation
- Cannot share a reservation with QUERY assignments
- Slots are allocated (and billed) as long as the reservation exists

````python
# Reservation type change pattern (reference only — not executed)
#
# To switch from QUERY to CONTINUOUS:
#
# Step 1: Delete the existing QUERY assignment
# reservation_client.delete_assignment(name=assignment.name)
#
# Step 2: Create a CONTINUOUS assignment
# continuous_assignment = reservation_client.create_assignment(
#     parent=reservation.name,
#     assignment=bigquery_reservation_v1.Assignment(
#         job_type=bigquery_reservation_v1.Assignment.JobType.CONTINUOUS,
#         assignee=f'projects/{PROJECT_ID}',
#     ),
# )
#
# Note: You may need a separate reservation for CONTINUOUS assignments
# since QUERY and CONTINUOUS cannot share the same reservation.

print('Reservation type change pattern:')
print('  1. Delete QUERY assignment')
print('  2. Create CONTINUOUS assignment (same or new reservation)')
print('  3. CONTINUOUS max: 500 slots per reservation')
print('  4. Cannot mix QUERY and CONTINUOUS in one reservation')
````

### The Continuous Query SQL

The key change is the `APPENDS()` function. It watches a table for newly inserted rows and triggers the query each time new data arrives. The query then re-computes features only for entities (users) that have new activity.

```sql
APPENDS(TABLE `project.dataset.table`, start_timestamp)
```

- `TABLE` — the source table to watch
- `start_timestamp` — only consider rows inserted after this time

The continuous query identifies entities with new rows, re-computes their features against the full history, and exports the updated features to Bigtable. This is efficient: only affected entities are reprocessed, but their features reflect the complete historical context.

````python
# Continuous query SQL (reference only — NOT executed)
# This query would run indefinitely, re-exporting features
# whenever new rows appear in the source table.

CONTINUOUS_SQL = f"""
-- =================================================================
-- CONTINUOUS QUERY VERSION
-- Triggered by APPENDS() — runs whenever new rows arrive.
-- Re-computes features for entities with new activity.
-- =================================================================
EXPORT DATA OPTIONS (
    format = 'CLOUD_BIGTABLE',
    uri = '{BIGTABLE_URI}',
    overwrite = false
) AS
WITH
-- Identify entities with new activity
new_activity AS (
    SELECT DISTINCT entity_id
    FROM APPENDS(
        TABLE `{BQ_PROJECT}.{BQ_DATASET}.sparse_events`,
        CURRENT_TIMESTAMP() - INTERVAL 10 MINUTE
    )
),

-- Re-compute features for active entities from full history
recomputed AS (
    SELECT
        d.entity_id,
        d.feature_timestamp,
        d.feature_1,
        d.feature_2,
        d.feature_3,
        d.feature_4,
        d.feature_5,
        ROW_NUMBER() OVER (
            PARTITION BY d.entity_id ORDER BY d.feature_timestamp DESC
        ) AS rn
    FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features` d
    WHERE d.entity_id IN (SELECT entity_id FROM new_activity)
),

latest_per_entity AS (
    SELECT * FROM recomputed WHERE rn = 1
)

SELECT
    CAST(entity_id AS STRING) AS rowkey,
    STRUCT(
        CAST(feature_timestamp AS STRING) AS _ts,
        CONCAT(
            CAST(feature_1 AS STRING), '|',
            CAST(feature_2 AS STRING), '|',
            CAST(feature_3 AS STRING), '|',
            CAST(feature_4 AS STRING), '|',
            CAST(feature_5 AS STRING)
        ) AS _values
    ) AS {BT_COLUMN_FAMILY}
FROM latest_per_entity
"""

print('Continuous query SQL (reference only \u2014 not executed):')
print(CONTINUOUS_SQL)
````

### How the Continuous Query Processes Data

```
Timeline:
  t=0    New row inserted into sparse_events for entity_42
  t=1s   APPENDS() detects the new row
  t=2s   CTE identifies entity_42 as having new activity
  t=3s   Features re-computed for entity_42 from full dense_features history
  t=5s   Updated features exported to Bigtable via EXPORT DATA
  t=5s   Bigtable now serves fresh features for entity_42
```

### When to Use Continuous Queries

| Scenario | Recommendation |
|----------|---------------|
| Features change hourly or slower | Scheduled query (cheaper) |
| Features need sub-minute freshness | Continuous query |
| Features need sub-second freshness | Streaming pipeline (Section 4) |
| Feature computation is simple (pass-through) | Continuous query works great |
| Feature computation is expensive (full window aggregation) | Continuous query works but uses more slots |

### Why We Don't Execute It Here

Continuous queries run indefinitely and consume reservation slots the entire time. For a learning exercise, the one-time export in Section 1 demonstrates the same EXPORT DATA mechanics. The upgrade path from one-time to continuous is:
1. Change reservation assignment from QUERY to CONTINUOUS
2. Wrap the source with `APPENDS()`
3. Add entity filtering in the CTE
4. Execute — the query runs until cancelled

## Section 4: Streaming Pipeline (Pub/Sub → Dataflow → Bigtable)

For the lowest possible latency, bypass BigQuery entirely for incoming events. A streaming pipeline computes features on the fly and writes directly to Bigtable.

In production, this pattern uses [Dataflow](https://cloud.google.com/dataflow/docs/overview) with the [Bigtable I/O connector](https://cloud.google.com/bigtable/docs/dataflow-bigtable). Below, we run an executable demo using a simpler Pub/Sub subscriber pattern in-notebook to illustrate the core event-processing logic — the same logic you would place inside a Dataflow `DoFn` or a Cloud Function.

### Architecture

```
                                        ┌─────────────────────────────┐
                                        │  Bigtable (feature serving) │
                                   ┌──→ │  - sub-second writes        │
                                   │    │  - millisecond reads         │
                                   │    └─────────────────────────────┘
                                   │
Events → Pub/Sub → Dataflow ───────┼────
                                   │
                                   │    ┌─────────────────────────────┐
                                   └──→ │  BigQuery (analytics)       │
                                        │  - historical analysis       │
                                        │  - model retraining          │
                                        └─────────────────────────────┘
```

### Dual-Write Pattern

The streaming pipeline writes to **two** destinations:
- **Bigtable** — for real-time feature serving (millisecond reads)
- **BigQuery** — for analytics, monitoring, and model retraining (via streaming insert or Storage Write API)

This avoids the latency of going through BigQuery first. Events hit Bigtable within seconds of arriving in Pub/Sub.

### Executable Demo: Pub/Sub → Bigtable

In production, streaming pipelines typically use [Dataflow](https://cloud.google.com/dataflow/docs/overview) with the [Bigtable I/O connector](https://cloud.google.com/bigtable/docs/dataflow-bigtable). Here, we demonstrate the core pattern using a Pub/Sub subscriber that writes directly to Bigtable — the same event-processing logic you'd put inside a Dataflow `DoFn` or a Cloud Function.

````python
# Create Pub/Sub topic and subscription for the streaming demo
from google.cloud import pubsub_v1

publisher = pubsub_v1.PublisherClient()
subscriber_client = pubsub_v1.SubscriberClient()

STREAM_TOPIC = 'feature-stream-demo'
STREAM_SUB = 'feature-stream-demo-sub'
topic_path = publisher.topic_path(PROJECT_ID, STREAM_TOPIC)
sub_path = subscriber_client.subscription_path(PROJECT_ID, STREAM_SUB)

try:
    publisher.create_topic(request={'name': topic_path})
    print(f'Created topic: {topic_path}')
except Exception as e:
    if 'ALREADY_EXISTS' in str(e):
        print(f'Topic exists: {topic_path}')
    else:
        raise

try:
    subscriber_client.create_subscription(
        request={'name': sub_path, 'topic': topic_path}
    )
    print(f'Created subscription: {sub_path}')
except Exception as e:
    if 'ALREADY_EXISTS' in str(e):
        print(f'Subscription exists: {sub_path}')
    else:
        raise
````

````python
# Publish 20 sample events (simulating a stream of feature updates)
import json, time, random

events = []
publish_start = time.perf_counter()

for i in range(20):
    event = {
        'entity_group': random.choice(['A', 'B', 'C']),
        'entity_id': str(random.randint(1, 100)).zfill(6),
        'event_type': random.choice(['purchase', 'click', 'view']),
        'metric_1': round(random.uniform(0, 100), 2),
        'metric_2': round(random.uniform(0, 50), 2),
        'timestamp': time.time(),
        'event_id': f'stream-evt-{i:03d}',
    }
    future = publisher.publish(topic_path, json.dumps(event).encode('utf-8'))
    future.result()
    events.append(event)

publish_time = (time.perf_counter() - publish_start) * 1000
print(f'Published {len(events)} events in {publish_time:.0f}ms ({publish_time/len(events):.1f}ms per event)')
````

````python
# Consume events and write each to Bigtable
consume_start = time.perf_counter()

response = subscriber_client.pull(
    request={'subscription': sub_path, 'max_messages': 50}
)
messages = response.received_messages
print(f'Pulled {len(messages)} messages')

# Write each event as a Bigtable row
STREAM_PREFIX = 'stream_demo#'
written_keys = []

for msg in messages:
    event = json.loads(msg.message.data.decode('utf-8'))
    
    # Row key: stream_demo#{entity_group}#{entity_id}#{event_id}
    row_key = f'{STREAM_PREFIX}{event["entity_group"]}#{event["entity_id"]}#{event["event_id"]}'
    
    row = table.direct_row(row_key.encode('utf-8'))
    row.set_cell(BT_COLUMN_FAMILY, 'event_type', event['event_type'].encode())
    row.set_cell(BT_COLUMN_FAMILY, 'metric_1', str(event['metric_1']).encode())
    row.set_cell(BT_COLUMN_FAMILY, 'metric_2', str(event['metric_2']).encode())
    row.set_cell(BT_COLUMN_FAMILY, 'timestamp', str(event['timestamp']).encode())
    row.commit()
    written_keys.append(row_key)

# Acknowledge messages
ack_ids = [msg.ack_id for msg in messages]
if ack_ids:
    subscriber_client.acknowledge(request={'subscription': sub_path, 'ack_ids': ack_ids})

consume_time = (time.perf_counter() - consume_start) * 1000
print(f'Wrote {len(written_keys)} rows to Bigtable in {consume_time:.0f}ms')
print(f'End-to-end: {publish_time + consume_time:.0f}ms (publish + consume + write)')
````

````python
# Verify: read back the streamed rows from Bigtable
from google.cloud.bigtable.row_set import RowSet, RowRange
from google.cloud.bigtable.row_filters import CellsColumnLimitFilter

row_set = RowSet()
row_set.add_row_range(RowRange(
    start_key=STREAM_PREFIX.encode(),
    end_key=f'{STREAM_PREFIX}\xff'.encode(),
))

verify_rows = list(table.read_rows(
    row_set=row_set,
    filter_=CellsColumnLimitFilter(1),
))

print(f'Verified: {len(verify_rows)} rows in Bigtable (expected {len(written_keys)})')
if verify_rows:
    sample = verify_rows[0]
    print(f'\nSample row: {sample.row_key.decode()}')
    for col, cells in sample.cells.get(BT_COLUMN_FAMILY, {}).items():
        print(f'  {col.decode()}: {cells[0].value.decode()}')
````

````python
# Cleanup: delete Pub/Sub resources and demo rows
try:
    subscriber_client.delete_subscription(request={'subscription': sub_path})
    print(f'Deleted subscription: {sub_path}')
except Exception as e:
    print(f'Subscription cleanup: {e}')

try:
    publisher.delete_topic(request={'topic': topic_path})
    print(f'Deleted topic: {topic_path}')
except Exception as e:
    print(f'Topic cleanup: {e}')

# Delete demo rows from Bigtable
deleted = 0
for row_key in written_keys:
    row = table.direct_row(row_key.encode('utf-8'))
    row.delete()
    row.commit()
    deleted += 1
print(f'Deleted {deleted} demo rows from Bigtable')
````

### Streaming Pipeline: When to Use

| Aspect | Continuous Query | Streaming Pipeline |
|--------|-----------------|-------------------|
| **Latency** | Seconds (BQ processing time) | Sub-second (direct writes) |
| **Feature logic** | Full SQL (windows, joins, UDFs) | Beam transforms (limited vs. SQL) |
| **Infrastructure** | BQ reservation only | Pub/Sub + Dataflow + Bigtable |
| **Operational complexity** | Low (managed by BQ) | High (pipeline monitoring, scaling, failures) |
| **Cost model** | Slot-hours | Dataflow worker-hours + Pub/Sub throughput |
| **Best for** | SQL-expressible features | Custom logic, lowest latency, dual-write patterns |

Most feature stores start with EXPORT DATA (Section 1) and only graduate to streaming when sub-second freshness is a hard requirement. The continuous query (Section 3) handles the middle ground well.

## Section 6: Reservation Management

BigQuery Enterprise edition uses [reservations](https://cloud.google.com/bigquery/docs/reservations-intro) to allocate compute capacity. Understanding reservation management is critical for controlling costs and ensuring feature export jobs have the resources they need.

### Key Concepts

| Concept | Description |
|---------|------------|
| **Reservation** | A pool of BigQuery slots (compute units) |
| **Assignment** | Maps a reservation to a project/folder/org and a job type (QUERY, CONTINUOUS, etc.) |
| **Slot** | A unit of BigQuery compute capacity |
| **Edition** | Enterprise, Enterprise Plus, or Standard — determines features and pricing |

### QUERY vs. CONTINUOUS Assignments

| Aspect | QUERY | CONTINUOUS |
|--------|-------|------------|
| **Purpose** | Ad-hoc queries, EXPORT DATA | Long-running continuous queries |
| **Lifecycle** | Jobs start and finish | Jobs run indefinitely |
| **Max slots** | Up to reservation capacity | 500 slots max per reservation |
| **Mixing** | Cannot share reservation with CONTINUOUS | Cannot share reservation with QUERY |
| **Cost** | Slots billed while reservation exists | Slots billed while reservation exists |

### Slot Capacity Planning

For EXPORT DATA to Bigtable:
- **Small tables** (< 1 GB): 100 slots is sufficient
- **Medium tables** (1-100 GB): 200-500 slots for reasonable export times
- **Large tables** (> 100 GB): 500-1000+ slots, consider partitioned exports

### Cost Implications

Enterprise edition on-demand pricing (as of 2024):
- Slots are billed per-second while the reservation exists
- Create the reservation, run the export, delete immediately — minimizes cost
- For continuous queries: slots are billed 24/7 since the reservation must stay active

## Section 7: Cleanup

Delete the reservation and assignment to stop slot billing. The features remain in Bigtable — only the BigQuery reservation resources are removed.

> **Important:** Delete in reverse order — assignment first, then reservation. The API rejects deletion of a reservation that still has assignments.

````python
# Delete reservation resources (reverse order)
print('Cleaning up reservation resources...')

# Step 1: Delete all assignments for the reservation
try:
    assignments = list(reservation_client.list_assignments(parent=reservation.name))
    for a in assignments:
        reservation_client.delete_assignment(name=a.name)
        print(f'  Deleted assignment: {a.name}')
except NotFound:
    print(f'  No assignments found')
except Exception as e:
    print(f'  Assignment cleanup: {e}')

# Step 2: Delete the reservation
try:
    reservation_client.delete_reservation(name=reservation.name)
    print(f'  Deleted reservation: {reservation.name}')
except NotFound:
    print(f'  Reservation already deleted')
except Exception as e:
    print(f'  Reservation cleanup: {e}')

print('\nWaiting 30s for deletion to propagate...')
time.sleep(30)
print('Reservation cleanup complete. Features remain in Bigtable.')
````

````python
# Verify cleanup
remaining = list(reservation_client.list_reservations(parent=parent))
our_reservations = [r for r in remaining if RESERVATION_ID in r.name]
if our_reservations:
    print(f'WARNING: Reservation still exists: {our_reservations[0].name}')
    print('  This will continue to incur slot charges.')
else:
    print(f'Reservation {RESERVATION_ID} successfully deleted.')
    print('No ongoing slot charges.')
````

## Summary

This notebook explored four synchronization strategies for keeping Bigtable features fresh, from simplest to most complex:

| Strategy | Freshness | Complexity | Cost | Best For |
|----------|-----------|------------|------|----------|
| **One-time EXPORT DATA** | Snapshot only | Lowest | Lowest | Backfill, schema migration |
| **Scheduled query** | Minutes to hours | Low | Low-moderate | Slowly-changing features |
| **Continuous query** | Seconds | Moderate | Moderate-high | SQL-expressible features needing near-real-time updates |
| **Streaming pipeline** | Sub-second | Highest | Variable | Lowest latency, custom logic |

### Key Takeaways

1. **Start simple.** One-time EXPORT DATA is the right first step. Add complexity only when freshness requirements demand it.

2. **Reservations control cost.** Create reservations on demand and delete them immediately after use. For continuous queries, plan for 24/7 slot costs.

3. **APPENDS() is the bridge.** The continuous query pattern reuses the same EXPORT DATA SQL — the only structural change is wrapping the source with `APPENDS()` and switching to a CONTINUOUS reservation.

4. **Streaming is a different paradigm.** When sub-second freshness is required, bypass BigQuery for the hot path entirely and write directly to Bigtable via Dataflow.
