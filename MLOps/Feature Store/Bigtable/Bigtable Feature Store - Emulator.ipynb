# Bigtable Feature Store — Emulator

The [Bigtable emulator](https://cloud.google.com/bigtable/docs/emulator) runs a local, in-memory Bigtable server for development and testing. No GCP project, no billing, no network — just a local process that speaks the Bigtable API.

This notebook demonstrates how to start the emulator, connect to it, and run feature store operations locally. It also covers the emulator's limitations and testing patterns for CI/CD pipelines.

**Prerequisites:** `gcloud` CLI installed (for the emulator binary). No GCP project or Bigtable instance required.

---
## Configuration

````python
import os
import subprocess
import sys
import time
import signal

EMULATOR_HOST = 'localhost'
EMULATOR_PORT = 8086
EMULATOR_ADDR = f'{EMULATOR_HOST}:{EMULATOR_PORT}'

# These are fake — the emulator doesn't validate them
EMU_PROJECT_ID = 'emulator-project'
EMU_INSTANCE_ID = 'emulator-instance'
EMU_TABLE_ID = 'features'
EMU_COLUMN_FAMILY = 'features'

print(f'Emulator address:  {EMULATOR_ADDR}')
print(f'Project (fake):    {EMU_PROJECT_ID}')
print(f'Instance (fake):   {EMU_INSTANCE_ID}')
````

---
## 1. What the Emulator Supports (and Doesn't)

The emulator implements a subset of the Bigtable API. Understanding what's missing is critical — tests that pass on the emulator may fail on production Bigtable if they rely on unsupported features.

| Feature | Emulator | Production | Impact on feature store testing |
|---------|----------|------------|--------------------------------|
| Read/write rows | Yes | Yes | Core CRUD operations work |
| Column families | Yes | Yes | Schema design testable |
| Row filters | Yes | Yes | Filtered reads work |
| Batch mutations | Yes | Yes | Batch writes testable |
| Check-and-mutate | Yes | Yes | Conditional writes work |
| Read-modify-write | Yes | Yes | Atomic operations work |
| **GoogleSQL** | **No** | Yes | KNN vector search (NB10) cannot be tested |
| **Admin API** | **No** | Yes | Cannot create/delete tables programmatically |
| **App profiles** | **No** | Yes | Cannot test routing or priority isolation |
| **IAM** | **No** | Yes | Cannot test access control |
| **Change streams** | **No** | Yes | Cannot test change stream consumers |
| **Data persistence** | **No** | Yes | Data lost on emulator restart |
| **Replication** | **No** | Yes | Cannot test multi-cluster behavior (NB11) |
| **Data Boost** | **No** | Yes | Cannot test serverless batch reads |

> **Key limitation:** The emulator does not support [GoogleSQL for Bigtable](https://cloud.google.com/bigtable/docs/googlesql-overview). This means SQL-based queries, including KNN vector search with `COSINE_DISTANCE()` and `TO_VECTOR32()` from [Notebook 10](./Bigtable%20Feature%20Store%20-%20Vector%20Storage%20and%20KNN%20Search.ipynb), cannot be tested locally.

## 2. Start the Emulator

````python
# Check if the emulator is available
result = subprocess.run(
    ['gcloud', 'components', 'list', '--filter=id=bigtable', '--format=value(state.name)'],
    capture_output=True, text=True,
)

if result.returncode != 0 or 'Installed' not in result.stdout:
    print('Bigtable emulator not available.')
    print('Install: gcloud components install bigtable')
    EMULATOR_AVAILABLE = False
else:
    EMULATOR_AVAILABLE = True
    print('Bigtable emulator available ✔')
````

````python
# Start the emulator as a background process
emulator_process = None

if EMULATOR_AVAILABLE:
    # Check if already running
    check = subprocess.run(
        ['lsof', '-i', f':{EMULATOR_PORT}'],
        capture_output=True, text=True,
    )
    
    if check.stdout.strip():
        print(f'Port {EMULATOR_PORT} already in use — emulator may already be running')
    else:
        emulator_process = subprocess.Popen(
            ['gcloud', 'beta', 'emulators', 'bigtable', 'start',
             f'--host-port={EMULATOR_ADDR}'],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
        )
        time.sleep(3)  # Wait for startup
        
        if emulator_process.poll() is None:
            print(f'Emulator started (PID {emulator_process.pid})')
            print(f'  Address: {EMULATOR_ADDR}')
        else:
            print(f'Emulator failed to start')
            stderr = emulator_process.stderr.read().decode()
            print(f'  Error: {stderr[:200]}')
            EMULATOR_AVAILABLE = False
    
    # Set the environment variable that tells the Bigtable client to use the emulator
    os.environ['BIGTABLE_EMULATOR_HOST'] = EMULATOR_ADDR
    print(f'  BIGTABLE_EMULATOR_HOST={EMULATOR_ADDR}')
else:
    print('Skipping emulator start — not available')
````

---
## 3. Connect and Use

When `BIGTABLE_EMULATOR_HOST` is set, the Bigtable Python client automatically connects to the emulator instead of production. No code changes needed — the same client code works against both.

````python
# Connect to the emulator — same API as production
if EMULATOR_AVAILABLE:
    from google.cloud import bigtable
    from google.cloud.bigtable.row_filters import CellsColumnLimitFilter
    
    emu_client = bigtable.Client(project=EMU_PROJECT_ID, admin=True)
    emu_instance = emu_client.instance(EMU_INSTANCE_ID)
    
    # Create table (emulator supports table creation via the client API)
    emu_table = emu_instance.table(EMU_TABLE_ID)
    
    try:
        emu_table.create(column_families={EMU_COLUMN_FAMILY: None, 'metadata': None})
        print(f'Created table: {EMU_TABLE_ID}')
    except Exception as e:
        if 'already exists' in str(e).lower():
            print(f'Table already exists: {EMU_TABLE_ID}')
        else:
            print(f'Table creation: {e}')
    
    print(f'Connected to emulator at {EMULATOR_ADDR}')
    print(f'  Project:  {EMU_PROJECT_ID} (fake — emulator ignores this)')
    print(f'  Instance: {EMU_INSTANCE_ID} (fake)')
    print(f'  Table:    {EMU_TABLE_ID}')
else:
    print('Emulator not available — skipping')
````

````python
# Write and read features — identical code to production notebooks
if EMULATOR_AVAILABLE:
    import json
    
    # Write sample entities
    entities = [
        ('A#000001', '29|4136|91|787331|25|21.3848|227.0087|20.2679|6.2978|0.3935'),
        ('A#000002', '730|9508|61|231617|315|48.99|173.0915|13.7933|2.9075|0.2936'),
        ('A#000003', '13|7614|2|907353|16|83.0991|118.6619|-15.6072|3.1276|0.6376'),
    ]
    
    for key, values in entities:
        row = emu_table.direct_row(key.encode('utf-8'))
        row.set_cell(EMU_COLUMN_FAMILY, '_values', values.encode('utf-8'))
        row.commit()
    
    # Write schema row
    schema = {
        'columns': ['feature_int_1', 'feature_int_2', 'feature_int_3', 'feature_int_4', 'feature_int_5',
                     'feature_float_1', 'feature_float_2', 'feature_float_3', 'feature_float_4', 'feature_float_5'],
        'delimiter': '|',
        'version': 'v1',
    }
    schema_row = emu_table.direct_row('#schema')
    schema_row.set_cell(EMU_COLUMN_FAMILY, '_schema', json.dumps(schema).encode('utf-8'))
    schema_row.commit()
    
    print(f'Wrote {len(entities)} entities + schema row to emulator')
    
    # Read back — same code as Notebook 1
    for key, _ in entities:
        row = emu_table.read_row(key.encode('utf-8'), filter_=CellsColumnLimitFilter(1))
        if row:
            values = row.cells[EMU_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
            print(f'  {key}: {values[:50]}...')
    
    print('\nThe same Python code works against both emulator and production.')
    print('The only difference is the BIGTABLE_EMULATOR_HOST environment variable.')
else:
    print('Emulator not available — skipping')
````

````python
# Demonstrate batch mutations and filtered reads on the emulator
if EMULATOR_AVAILABLE:
    # Batch write 20 more entities
    rows_to_write = []
    for i in range(4, 24):
        key = f'A#{str(i).zfill(6)}'
        values = f'{i}|{i*100}|{i*10}|{i*1000}|{i}|{i*1.5:.4f}|{i*2.5:.4f}|0.0|0.0|{i*0.01:.4f}'
        row = emu_table.direct_row(key.encode('utf-8'))
        row.set_cell(EMU_COLUMN_FAMILY, '_values', values.encode('utf-8'))
        rows_to_write.append(row)
    
    statuses = emu_table.mutate_rows(rows_to_write)
    print(f'Batch wrote {len(rows_to_write)} rows')
    
    # Read all rows with a filter
    all_rows = list(emu_table.read_rows(filter_=CellsColumnLimitFilter(1)))
    data_rows = [r for r in all_rows if not r.row_key.decode().startswith('#')]
    print(f'Total data rows: {len(data_rows)}')
    
    # Prefix scan — same as NB6
    from google.cloud.bigtable.row_set import RowSet, RowRange
    
    row_set = RowSet()
    row_set.add_row_range(RowRange(
        start_key=b'A#000001',
        end_key=b'A#000005',
    ))
    
    prefix_rows = list(emu_table.read_rows(
        row_set=row_set,
        filter_=CellsColumnLimitFilter(1),
    ))
    print(f'Prefix scan (A#000001..A#000005): {len(prefix_rows)} rows')
    
    for row in prefix_rows:
        print(f'  {row.row_key.decode()}')
else:
    print('Emulator not available — skipping')
````

## 4. Testing Patterns

### pytest Fixture for Emulator Lifecycle

```python
import pytest
import subprocess
import os
import time

@pytest.fixture(scope='session')
def bigtable_emulator():
    """Start Bigtable emulator for the test session."""
    process = subprocess.Popen(
        ['gcloud', 'beta', 'emulators', 'bigtable', 'start',
         '--host-port=localhost:8086'],
        stdout=subprocess.PIPE, stderr=subprocess.PIPE,
    )
    os.environ['BIGTABLE_EMULATOR_HOST'] = 'localhost:8086'
    time.sleep(3)
    
    yield process
    
    process.terminate()
    process.wait()
    del os.environ['BIGTABLE_EMULATOR_HOST']

@pytest.fixture
def feature_table(bigtable_emulator):
    """Create a fresh feature table for each test."""
    from google.cloud import bigtable
    client = bigtable.Client(project='test-project', admin=True)
    instance = client.instance('test-instance')
    table = instance.table('test-features')
    table.create(column_families={'features': None})
    
    yield table
    
    table.delete()

def test_write_and_read(feature_table):
    row = feature_table.direct_row(b'entity_001')
    row.set_cell('features', '_values', b'1|2|3')
    row.commit()
    
    result = feature_table.read_row(b'entity_001')
    assert result is not None
    assert result.cells['features'][b'_values'][0].value == b'1|2|3'
```

## 5. Limitations in Practice

### What You Can Test on the Emulator

| Test type | Works? | Notes |
|-----------|--------|-------|
| Feature encoding/decoding | **Yes** | Pipe-delimited, JSON, protobuf — all work |
| Write → read round-trips | **Yes** | Core correctness testing |
| Batch mutations | **Yes** | `mutate_rows()` works |
| Row filters | **Yes** | `CellsColumnLimitFilter`, `TimestampRangeFilter`, etc. |
| Atomic operations | **Yes** | `read_modify_write`, `check_and_mutate` |
| Schema versioning | **Yes** | Write/read versioned schema rows |

### What You Cannot Test

| Feature | Workaround |
|---------|-----------|
| **GoogleSQL / KNN** | Skip these tests on emulator; run against a dedicated test instance |
| **App profile routing** | Test routing logic in integration tests against production |
| **Replication** | Requires Production instance — test in staging environment |
| **Data Boost** | Test scan logic on emulator; verify Data Boost profile in staging |
| **IAM / access control** | Test in staging with service account impersonation |
| **Performance / latency** | Emulator has different performance characteristics than production |

### Conditional Test Skipping

    # This test works on both emulator and production

## Cleanup

Stop the emulator process. Since the emulator is in-memory, all data is lost when it stops — no resources to clean up.

````python
# Stop the emulator
if EMULATOR_AVAILABLE:
    if emulator_process and emulator_process.poll() is None:
        emulator_process.terminate()
        emulator_process.wait(timeout=5)
        print(f'Stopped emulator (PID {emulator_process.pid})')
    else:
        print('Emulator process not managed by this notebook — stop it manually')
    
    # Remove the environment variable
    if 'BIGTABLE_EMULATOR_HOST' in os.environ:
        del os.environ['BIGTABLE_EMULATOR_HOST']
        print('Cleared BIGTABLE_EMULATOR_HOST')
    
    print('\nAll emulator data has been discarded (in-memory only).')
    print('No GCP resources were created or modified.')
else:
    print('Emulator was not started — nothing to clean up')
````

## Summary

The Bigtable emulator provides a local, zero-cost environment for developing and testing feature store code:

| Aspect | Emulator | Production |
|--------|----------|------------|
| **Cost** | Free | Per-node-hour |
| **Setup** | `gcloud beta emulators bigtable start` | Instance creation, IAM setup |
| **Latency** | Local (< 1ms) | Network round-trip (2-8ms) |
| **Data persistence** | None (in-memory) | Durable (replicated) |
| **API coverage** | Partial (no GoogleSQL, no admin) | Full |
| **Best for** | Unit tests, CI/CD, local development | Integration tests, production serving |

### Decision Guide

```
What are you testing?
│
├─ Feature encoding/decoding logic ──→ Emulator (fast, free)
├─ Read/write correctness ───────────→ Emulator (fast, free)
├─ GoogleSQL queries ────────────────→ Production test instance
├─ App profile routing ──────────────→ Production test instance
├─ Replication behavior ─────────────→ Production test instance (NB11)
└─ End-to-end latency ──────────────→ Production (emulator latency is meaningless)
```
