## 1. Schema Versioning

In Notebook 1, we introduced the `#schema` metadata row — a single JSON blob that describes the feature columns, their types, and the delimiter. The serving client reads this row once at startup and builds a decoder.

That works for a static schema. But in production, schemas change:
- New features get added (a retrained model needs an extra signal)
- Features get deprecated (an old signal is no longer predictive)
- Types change (an INT64 becomes a FLOAT64 after reprocessing)

We need **versioned schemas** so that:
1. The serving client always knows which schema is current
2. Old schemas are preserved for debugging and rollback
3. The transition from v1 to v2 is atomic — no partial state

### Schema Registry Pattern

We use three kinds of schema rows:

| Row key | Purpose |
|---------|--------|
| `#schema#v1` | Version 1 schema (immutable once written) |
| `#schema#v2` | Version 2 schema (immutable once written) |
| `#schema#current` | Pointer to the active version — this is what the serving client reads |

The serving client reads `#schema#current` to discover the active version, then reads `#schema#vN` for the full schema definition. Updating the schema is a two-step process: write the new version row, then atomically update `#schema#current`.

````python
# Define v1 schema (the original schema from Notebook 1)
SCHEMA_V1 = {
    'version': 'v1',
    'columns': [
        'txn_count_5min',
        'txn_count_1hr',
        'spending_1hr',
        'spending_30d_avg',
        'spending_ratio',
        'category_count_24hr',
        'sale_price',
        'order_total',
        'account_age_days',
        'user_lifetime_orders',
    ],
    'types': {
        'txn_count_5min': 'INT64',
        'txn_count_1hr': 'INT64',
        'spending_1hr': 'FLOAT64',
        'spending_30d_avg': 'FLOAT64',
        'spending_ratio': 'FLOAT64',
        'category_count_24hr': 'INT64',
        'sale_price': 'FLOAT64',
        'order_total': 'FLOAT64',
        'account_age_days': 'INT64',
        'user_lifetime_orders': 'INT64',
    },
    'delimiter': '|',
    'created_at': datetime.datetime.now(datetime.timezone.utc).isoformat(),
    'description': 'Original behavioral fraud features from Notebook 1',
}

# Write #schema#v1
v1_json = json.dumps(SCHEMA_V1, indent=2)
row_v1 = table.direct_row('#schema#v1')
row_v1.set_cell(BT_COLUMN_FAMILY, '_schema', v1_json.encode('utf-8'))
row_v1.commit()
print(f'Wrote #schema#v1 ({len(v1_json)} bytes)')
print(f'  Columns: {len(SCHEMA_V1["columns"])}')
print(f'  Version: {SCHEMA_V1["version"]}')
````

````python
# Define v2 schema: adds two new features for the retrained model
SCHEMA_V2 = {
    'version': 'v2',
    'columns': [
        'txn_count_5min',
        'txn_count_1hr',
        'spending_1hr',
        'spending_30d_avg',
        'spending_ratio',
        'category_count_24hr',
        'sale_price',
        'order_total',
        'account_age_days',
        'user_lifetime_orders',
        # New in v2:
        'avg_order_gap_hours',
        'distinct_devices_7d',
    ],
    'types': {
        'txn_count_5min': 'INT64',
        'txn_count_1hr': 'INT64',
        'spending_1hr': 'FLOAT64',
        'spending_30d_avg': 'FLOAT64',
        'spending_ratio': 'FLOAT64',
        'category_count_24hr': 'INT64',
        'sale_price': 'FLOAT64',
        'order_total': 'FLOAT64',
        'account_age_days': 'INT64',
        'user_lifetime_orders': 'INT64',
        'avg_order_gap_hours': 'FLOAT64',
        'distinct_devices_7d': 'INT64',
    },
    'delimiter': '|',
    'created_at': datetime.datetime.now(datetime.timezone.utc).isoformat(),
    'description': 'Added avg_order_gap_hours and distinct_devices_7d for coordinated fraud detection',
    'parent_version': 'v1',
}

# Write #schema#v2
v2_json = json.dumps(SCHEMA_V2, indent=2)
row_v2 = table.direct_row('#schema#v2')
row_v2.set_cell(BT_COLUMN_FAMILY, '_schema', v2_json.encode('utf-8'))
row_v2.commit()
print(f'Wrote #schema#v2 ({len(v2_json)} bytes)')
print(f'  Columns: {len(SCHEMA_V2["columns"])}')
print(f'  New:     avg_order_gap_hours, distinct_devices_7d')
print(f'  Parent:  v1')
````

````python
# Write #schema#current — points to v1 initially
current_pointer = json.dumps({'current_version': 'v1'})
row_current = table.direct_row('#schema#current')
row_current.set_cell(BT_COLUMN_FAMILY, '_schema', current_pointer.encode('utf-8'))
row_current.commit()
print(f'Wrote #schema#current -> v1')

# Read it back to verify
current_row = table.read_row('#schema#current', filter_=CellsColumnLimitFilter(1))
pointer = json.loads(current_row.cells[BT_COLUMN_FAMILY][b'_schema'][0].value.decode('utf-8'))
print(f'Verified: current_version = {pointer["current_version"]}')
````

### Schema-Aware Reader

The serving client uses a two-step read: first `#schema#current` to discover the active version, then `#schema#vN` for the full schema definition. This function encapsulates the pattern.

````python
TYPE_PARSERS = {'INT64': int, 'FLOAT64': float, 'STRING': str}

def load_current_schema(tbl, family):
    """Read #schema#current and return the active schema dict."""
    # Step 1: Read the pointer
    current_row = tbl.read_row('#schema#current', filter_=CellsColumnLimitFilter(1))
    pointer = json.loads(current_row.cells[family][b'_schema'][0].value.decode('utf-8'))
    version = pointer['current_version']
    
    # Step 2: Read the versioned schema
    schema_row = tbl.read_row(f'#schema#{version}', filter_=CellsColumnLimitFilter(1))
    schema = json.loads(schema_row.cells[family][b'_schema'][0].value.decode('utf-8'))
    return schema

def decode_features(raw_value, schema):
    """Decode a delimited feature string using the schema."""
    parts = raw_value.split(schema['delimiter'])
    result = {}
    for i, col in enumerate(schema['columns']):
        if i < len(parts):
            result[col] = TYPE_PARSERS[schema['types'][col]](parts[i])
        else:
            # Missing value — backward compatibility for new columns
            result[col] = None
    return result

# Test: load current schema and decode a feature row
active_schema = load_current_schema(table, BT_COLUMN_FAMILY)
print(f'Active schema version: {active_schema["version"]}')
print(f'Columns: {len(active_schema["columns"])}')
print(f'Description: {active_schema["description"]}')
````

### Switching Versions

Upgrading the schema is a single row write — update `#schema#current` to point to the new version. The serving client picks up the change on its next schema refresh.

````python
def update_current_schema(tbl, family, new_version):
    """Atomically switch #schema#current to a new version."""
    pointer = json.dumps({'current_version': new_version})
    row = tbl.direct_row('#schema#current')
    row.set_cell(family, '_schema', pointer.encode('utf-8'))
    row.commit()
    return new_version

# Upgrade to v2
update_current_schema(table, BT_COLUMN_FAMILY, 'v2')
schema_after = load_current_schema(table, BT_COLUMN_FAMILY)
print(f'Schema upgraded: v1 -> {schema_after["version"]}')
print(f'New columns: {[c for c in schema_after["columns"] if c not in SCHEMA_V1["columns"]]}')

# Roll back to v1 for the rest of the notebook (since data rows are v1 format)
update_current_schema(table, BT_COLUMN_FAMILY, 'v1')
print(f'\nRolled back to v1 for compatibility with existing data rows.')
````

### Interpreting Schema Versioning

The versioned schema pattern gives you three properties:

1. **Immutable history.** Each `#schema#vN` row is written once and never modified. You can always see what the schema looked like at any point in time.

2. **Atomic switch.** Updating `#schema#current` is a single Bigtable row write — atomic and fast. There is no window where the schema is "half-upgraded."

3. **Instant rollback.** If the new schema causes issues, point `#schema#current` back to the previous version. The data rows are unchanged.

The trade-off is that your decoder must handle rows written under any schema version — old rows may have fewer fields than the current schema expects. We address that next.

## 2. Adding Features (Non-Breaking)

Adding a new column qualifier to Bigtable is effortless — just start writing it. There is no `ALTER TABLE ADD COLUMN`. Bigtable is schemaless at the storage level; columns are created on first write.

The challenge is on the **read side**: old rows do not have the new qualifier. Your decoder must handle missing values gracefully.

### Pattern: Append to Delimited Payload

For the concatenated `_values` payload, new features are appended to the end of the delimiter-separated string. Old decoders that only know about the first N fields still work — they parse the fields they know and ignore the rest. New decoders parse all fields.

For rows that haven't been backfilled yet, the new fields are simply missing from the string. The decoder returns `None` for those fields.

````python
# Demo: Write some rows with the v1 format (10 features)
# and some with the v2 format (12 features, 2 new appended)

import random
random.seed(42)

# Write 5 rows in v1 format
v1_rows = []
for i in range(5):
    row_key = f'demo_v1_{i}'
    values = [
        str(random.randint(0, 3)),       # txn_count_5min
        str(random.randint(0, 10)),      # txn_count_1hr
        f'{random.uniform(0, 500):.2f}', # spending_1hr
        f'{random.uniform(20, 100):.2f}',# spending_30d_avg
        f'{random.uniform(0.1, 5):.4f}', # spending_ratio
        str(random.randint(0, 5)),       # category_count_24hr
        f'{random.uniform(10, 300):.2f}',# sale_price
        f'{random.uniform(10, 300):.2f}',# order_total
        str(random.randint(30, 3000)),   # account_age_days
        str(random.randint(1, 50)),      # user_lifetime_orders
    ]
    payload = '|'.join(values)
    row = table.direct_row(row_key)
    row.set_cell(BT_COLUMN_FAMILY, '_values', payload.encode('utf-8'))
    row.commit()
    v1_rows.append(row_key)

# Write 5 rows in v2 format (v1 fields + 2 new fields appended)
v2_rows = []
for i in range(5):
    row_key = f'demo_v2_{i}'
    values = [
        str(random.randint(0, 3)),
        str(random.randint(0, 10)),
        f'{random.uniform(0, 500):.2f}',
        f'{random.uniform(20, 100):.2f}',
        f'{random.uniform(0.1, 5):.4f}',
        str(random.randint(0, 5)),
        f'{random.uniform(10, 300):.2f}',
        f'{random.uniform(10, 300):.2f}',
        str(random.randint(30, 3000)),
        str(random.randint(1, 50)),
        # New v2 fields:
        f'{random.uniform(1, 720):.1f}',  # avg_order_gap_hours
        str(random.randint(1, 5)),         # distinct_devices_7d
    ]
    payload = '|'.join(values)
    row = table.direct_row(row_key)
    row.set_cell(BT_COLUMN_FAMILY, '_values', payload.encode('utf-8'))
    row.commit()
    v2_rows.append(row_key)

print(f'Wrote {len(v1_rows)} v1 rows (10 fields) and {len(v2_rows)} v2 rows (12 fields)')
````

````python
# Read all demo rows using the v2 schema — the decoder handles missing fields
v2_schema = load_current_schema(table, BT_COLUMN_FAMILY)
# Temporarily use v2 for decoding
update_current_schema(table, BT_COLUMN_FAMILY, 'v2')
v2_schema = load_current_schema(table, BT_COLUMN_FAMILY)

print(f'Decoding all demo rows with v2 schema ({len(v2_schema["columns"])} columns):\n')

all_demo_keys = v1_rows + v2_rows
for key in all_demo_keys:
    row = table.read_row(key, filter_=CellsColumnLimitFilter(1))
    if row and b'_values' in row.cells.get(BT_COLUMN_FAMILY, {}):
        raw = row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
        features = decode_features(raw, v2_schema)
        new_fields = {
            'avg_order_gap_hours': features.get('avg_order_gap_hours'),
            'distinct_devices_7d': features.get('distinct_devices_7d'),
        }
        print(f'  {key:<12s}  new_fields={new_fields}')

# Restore to v1
update_current_schema(table, BT_COLUMN_FAMILY, 'v1')
print(f'\nRows with v1 data return None for new fields — backward compatible.')
````

### Handling Default Values

When the decoder returns `None` for a missing feature, the serving application needs a policy:

| Strategy | When to use |
|----------|------------|
| **Default value** (e.g., 0, -1, mean) | The model tolerates a sentinel value without degrading predictions |
| **Skip the feature** | The model can handle missing inputs (e.g., tree models with `enable_missing=True`) |
| **Block serving** | The feature is critical and rows without it must not be scored |
| **Lazy backfill** | On read, compute the missing feature on-the-fly and write it back |

For our fraud model, `avg_order_gap_hours` defaults to a high value (infrequent ordering) and `distinct_devices_7d` defaults to 1. These are conservative defaults that reduce false positives during the backfill window.

## 3. Removing and Renaming Features

Removing a feature from a Bigtable table is different from removing a column in a relational database. There is no `DROP COLUMN` — old data retains the column qualifier until you explicitly delete it or GC policies reclaim it.

### Deprecation Pattern

Rather than immediately deleting data, we use a **deprecation workflow**:

1. **Mark as deprecated in schema metadata.** The schema row records which features are deprecated and when.
2. **Stop writing the feature.** New exports/updates omit it.
3. **Update readers.** The decoder ignores deprecated columns.
4. **Eventually clean up.** Backfill to remove old data, or let GC policies expire it.

````python
# Demonstrate deprecation: mark feature_int_5 (spending_ratio) as deprecated in v3
SCHEMA_V3 = {
    'version': 'v3',
    'columns': [
        'txn_count_5min',
        'txn_count_1hr',
        'spending_1hr',
        'spending_30d_avg',
        'spending_ratio',          # still in the column list for backward compat
        'category_count_24hr',
        'sale_price',
        'order_total',
        'account_age_days',
        'user_lifetime_orders',
    ],
    'types': {
        'txn_count_5min': 'INT64',
        'txn_count_1hr': 'INT64',
        'spending_1hr': 'FLOAT64',
        'spending_30d_avg': 'FLOAT64',
        'spending_ratio': 'FLOAT64',
        'category_count_24hr': 'INT64',
        'sale_price': 'FLOAT64',
        'order_total': 'FLOAT64',
        'account_age_days': 'INT64',
        'user_lifetime_orders': 'INT64',
    },
    'delimiter': '|',
    'created_at': datetime.datetime.now(datetime.timezone.utc).isoformat(),
    'description': 'Deprecated spending_ratio — redundant with spending_1hr/spending_30d_avg',
    'parent_version': 'v1',
    'deprecated': ['spending_ratio'],
    'deprecated_date': datetime.date.today().isoformat(),
}

# Write #schema#v3
v3_json = json.dumps(SCHEMA_V3, indent=2)
row_v3 = table.direct_row('#schema#v3')
row_v3.set_cell(BT_COLUMN_FAMILY, '_schema', v3_json.encode('utf-8'))
row_v3.commit()

print(f'Wrote #schema#v3 ({len(v3_json)} bytes)')
print(f'  Deprecated: {SCHEMA_V3["deprecated"]}')
print(f'  Date:       {SCHEMA_V3["deprecated_date"]}')
````

````python
# Deprecation-aware decoder
def decode_features_v2(raw_value, schema):
    """Decode features, excluding deprecated columns."""
    parts = raw_value.split(schema['delimiter'])
    deprecated = set(schema.get('deprecated', []))
    result = {}
    for i, col in enumerate(schema['columns']):
        if col in deprecated:
            continue  # Skip deprecated features
        if i < len(parts):
            result[col] = TYPE_PARSERS[schema['types'][col]](parts[i])
        else:
            result[col] = None
    return result

# Test: decode a demo row with v3 schema (spending_ratio excluded)
row = table.read_row('demo_v1_0', filter_=CellsColumnLimitFilter(1))
raw = row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')

features_full = decode_features(raw, SCHEMA_V3)
features_no_deprecated = decode_features_v2(raw, SCHEMA_V3)

print('Full decode (all columns):')
for k, v in features_full.items():
    marker = ' [DEPRECATED]' if k in SCHEMA_V3.get('deprecated', []) else ''
    print(f'  {k:<25s} = {v}{marker}')

print(f'\nDeprecation-aware decode ({len(features_no_deprecated)} columns):')
for k, v in features_no_deprecated.items():
    print(f'  {k:<25s} = {v}')
````

### Renaming Features

Renaming is effectively a deprecation of the old name plus addition of the new name. During the transition period, **both** exist in the data:

1. Write the new qualifier with the new name
2. Update the schema: add the new name, mark the old as deprecated
3. Run a backfill to copy old-name values to new-name values
4. Once all rows have the new name, remove the old name from the schema

This is inherently a multi-step process. The schema metadata tracks where you are in the transition:

```json
{
    "renames": {
        "spending_ratio": "price_to_avg_ratio"
    },
    "rename_status": "in_progress"
}
```

The serving client can use either name during the transition, falling back to the old name if the new one is not yet present.

## 4. Backfilling Without Downtime

When you add a new feature or recompute existing values, you need to update potentially hundreds of thousands of rows. The key constraint: **readers must continue serving during the backfill.** No downtime, no inconsistent reads.

### Backfill Strategies

| Pattern | How it works | Best for |
|---------|-------------|----------|
| **Write-alongside** | Write new values in a new qualifier; reader switches once backfill completes | Adding a feature to existing rows |
| **Batch MutateRows** | Update rows in batches using `mutate_rows()` | Recomputing values in-place |
| **EXPORT DATA overwrite** | Re-export the entire table from BigQuery with `overwrite=true` | Full table refresh |

### Demo: Batch Backfill with MutateRows

Let's backfill the `avg_order_gap_hours` and `distinct_devices_7d` features onto 1000 rows. We use `mutate_rows()` for efficient batching.

````python
# Read a sample of existing data rows (non-schema, non-demo rows)
partial_rows = table.read_rows(
    filter_=CellsColumnLimitFilter(1),
    limit=1100,
)
partial_rows.consume_all()

# Filter to data rows only (exclude schema and demo rows)
data_keys = [
    key for key in partial_rows.rows.keys()
    if not key.startswith(b'#') and not key.startswith(b'demo_')
]

print(f'Found {len(data_keys)} data rows available for backfill')

# Take up to 1000 for the demo
backfill_keys = data_keys[:1000]
print(f'Will backfill {len(backfill_keys)} rows')
````

````python
# Backfill: append two new features to each row's _values payload
import random
random.seed(123)

BATCH_SIZE = 100
total_updated = 0
total_errors = 0
start_time = time.time()

for batch_start in range(0, len(backfill_keys), BATCH_SIZE):
    batch_keys = backfill_keys[batch_start:batch_start + BATCH_SIZE]
    rows_to_mutate = []
    
    for key in batch_keys:
        # Read current value
        existing = partial_rows.rows.get(key)
        if existing and b'_values' in existing.cells.get(BT_COLUMN_FAMILY, {}):
            old_value = existing.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
            
            # Compute new feature values (in production, these come from BigQuery or a pipeline)
            avg_gap = f'{random.uniform(1, 720):.1f}'
            devices = str(random.randint(1, 5))
            
            # Append new fields to the existing payload
            new_value = f'{old_value}|{avg_gap}|{devices}'
            
            row = table.direct_row(key)
            row.set_cell(BT_COLUMN_FAMILY, '_values', new_value.encode('utf-8'))
            rows_to_mutate.append(row)
    
    # Write batch
    if rows_to_mutate:
        statuses = table.mutate_rows(rows_to_mutate)
        batch_errors = sum(1 for s in statuses if s.code != 0)
        total_updated += len(rows_to_mutate) - batch_errors
        total_errors += batch_errors
    
    # Progress report
    elapsed = time.time() - start_time
    done = min(batch_start + BATCH_SIZE, len(backfill_keys))
    rate = done / elapsed if elapsed > 0 else 0
    remaining = (len(backfill_keys) - done) / rate if rate > 0 else 0
    print(f'  Batch {batch_start // BATCH_SIZE + 1}: '
          f'{done}/{len(backfill_keys)} rows '
          f'({elapsed:.1f}s elapsed, ~{remaining:.0f}s remaining)')

total_time = time.time() - start_time
print(f'\nBackfill complete:')
print(f'  Updated: {total_updated} rows')
print(f'  Errors:  {total_errors} rows')
print(f'  Time:    {total_time:.1f}s')
print(f'  Rate:    {total_updated / total_time:.0f} rows/sec')
````

````python
# Verify: read a backfilled row and decode with v2 schema
if backfill_keys:
    sample_key = backfill_keys[0]
    row = table.read_row(sample_key, filter_=CellsColumnLimitFilter(1))
    raw = row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
    
    # Decode with v2 schema
    features = decode_features(raw, SCHEMA_V2)
    
    print(f'Verified backfilled row: {sample_key}')
    print(f'  Raw: {raw}')
    print(f'  Fields: {len(features)}')
    print(f'  avg_order_gap_hours: {features.get("avg_order_gap_hours")}')
    print(f'  distinct_devices_7d: {features.get("distinct_devices_7d")}')
````

### Zero-Downtime Backfill Flow

The key insight is that **readers and writers operate on different schema versions simultaneously**:

```
Time →

1. Write #schema#v2 (new version with new features)
2. Start backfill (batch MutateRows, appending new fields)
   ├─ Readers still using v1 schema — they ignore extra fields
   ├─ New rows being written with v2 format by the export pipeline
   └─ Old rows being updated one batch at a time
3. Backfill complete — all rows have v2 data
4. Update #schema#current -> v2 (atomic switch)
   └─ Readers now decode v2 fields
```

At no point during this process do readers see inconsistent data. They either read with the old schema (missing new fields default to `None`) or the new schema (all fields present).

### EXPORT DATA Alternative

For a full table refresh, [`EXPORT DATA`](https://cloud.google.com/bigquery/docs/export-to-bigtable) `... overwrite = true` from BigQuery replaces all rows atomically. This is simpler than batch mutations when:
- You are adding many features at once
- The export query computes all features (old and new) from the source table
- The table is small enough that a full re-export is faster than row-by-row updates

For our ~130K-row feature store, a full EXPORT DATA takes about 15-20 seconds. The trade-off is that `overwrite = true` briefly deletes old data before writing new data — there is a short window where reads may return empty. For truly zero-downtime refreshes, use batch mutations.

### Query Metrics via API

The Cloud Monitoring API lets you query these metrics programmatically — useful for dashboards, alerting, and capacity planning.

````python
# Query Bigtable read latencies from the last hour
monitoring_client = monitoring_v3.MetricServiceClient()
project_name = f'projects/{PROJECT_ID}'

now = time.time()
seconds_ago = int(now) - 3600  # 1 hour ago

# Build the time interval
interval = monitoring_v3.TimeInterval({
    'end_time': {'seconds': int(now)},
    'start_time': {'seconds': seconds_ago},
})

# Query server request count
try:
    results = monitoring_client.list_time_series(
        request={
            'name': project_name,
            'filter': (
                'metric.type = "bigtable.googleapis.com/server/request_count" '
                f'AND resource.labels.instance = "{BT_INSTANCE}"'
            ),
            'interval': interval,
            'view': monitoring_v3.ListTimeSeriesRequest.TimeSeriesView.FULL,
        }
    )
    
    total_requests = 0
    methods = {}
    for ts in results:
        method = ts.metric.labels.get('method', 'unknown')
        for point in ts.points:
            count = point.value.int64_value
            total_requests += count
            methods[method] = methods.get(method, 0) + count
    
    print(f'Bigtable request count (last hour):')
    print(f'  Instance: {BT_INSTANCE}')
    print(f'  Total:    {total_requests:,} requests')
    for method, count in sorted(methods.items(), key=lambda x: -x[1]):
        print(f'  {method:<25s} {count:>8,}')

except Exception as e:
    print(f'Monitoring query returned no data (expected if instance is new or idle).')
    print(f'Detail: {e}')
````

### Feature Distribution Monitoring

Infrastructure metrics (latency, CPU, storage) tell you if the system is healthy. **Feature distribution monitoring** tells you if the *data* is healthy. Feature drift — a statistical change in feature values between training and serving — can silently degrade model performance even when the infrastructure is running perfectly.

| Drift Type | What Changed | Example |
|-----------|-------------|---------|
| **Data drift** | Input feature distribution shifted | User age distribution skewed younger after a marketing campaign |
| **Concept drift** | Relationship between features and target changed | Purchase patterns shifted during a holiday season |
| **Schema drift** | Feature schema changed | A column switched from INT64 to FLOAT64 after an upstream change |

Catching drift early lets you retrain before model performance degrades.

````python
# Feature drift detection: compare Bigtable serving values to BigQuery baseline
from google.cloud.bigtable.row_filters import CellsColumnLimitFilter
import json
import numpy as np

# Step 1: Compute baseline statistics from BigQuery (training data)
print('Computing baseline statistics from BigQuery...')
baseline_df = bq.query(f'''
    SELECT
        AVG(feature_float_1) AS mean_f1,
        STDDEV(feature_float_1) AS std_f1,
        AVG(feature_float_2) AS mean_f2,
        STDDEV(feature_float_2) AS std_f2,
        AVG(feature_float_3) AS mean_f3,
        STDDEV(feature_float_3) AS std_f3,
        AVG(CAST(feature_int_1 AS FLOAT64)) AS mean_i1,
        STDDEV(CAST(feature_int_1 AS FLOAT64)) AS std_i1,
        AVG(CAST(feature_int_2 AS FLOAT64)) AS mean_i2,
        STDDEV(CAST(feature_int_2 AS FLOAT64)) AS std_i2,
    FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
''').to_dataframe()

# Step 2: Sample features from Bigtable (serving data)
print('Sampling features from Bigtable...')
schema_row = table.read_row('#schema', filter_=CellsColumnLimitFilter(1))
if schema_row is None:
    print('No #schema row — run Notebook 1 first to export features')
else:
    schema_data = json.loads(
        schema_row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
    )
    separator = schema_data.get('separator', schema_data.get('delimiter', '|'))
    columns = schema_data['columns']

    # Read a sample of rows
    partial_rows = table.read_rows(filter_=CellsColumnLimitFilter(1))
    sample_values = {col: [] for col in ['feature_float_1', 'feature_float_2', 'feature_float_3',
                                          'feature_int_1', 'feature_int_2']}
    count = 0
    for row in partial_rows:
        key = row.row_key.decode()
        if key.startswith('#'):
            continue
        raw = row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
        parts = raw.split(separator)
        for col in sample_values:
            if col in columns:
                idx = columns.index(col)
                if idx < len(parts):
                    try:
                        sample_values[col].append(float(parts[idx]))
                    except (ValueError, TypeError):
                        pass
        count += 1
        if count >= 500:
            break

    # Step 3: Compare distributions
    print(f'\n=== Feature Drift Report ({count} sampled rows) ===\n')
    print(f'{"Feature":<20} {"BQ Mean":>10} {"BQ Std":>10} {"BT Mean":>10} {"BT Std":>10} {"Shift (σ)":>10} {"Status":<8}')
    print('-' * 78)

    features_to_check = [
        ('feature_float_1', 'mean_f1', 'std_f1'),
        ('feature_float_2', 'mean_f2', 'std_f2'),
        ('feature_float_3', 'mean_f3', 'std_f3'),
        ('feature_int_1', 'mean_i1', 'std_i1'),
        ('feature_int_2', 'mean_i2', 'std_i2'),
    ]

    drift_detected = False
    for feat_name, mean_col, std_col in features_to_check:
        bq_mean = baseline_df[mean_col].iloc[0]
        bq_std = baseline_df[std_col].iloc[0]
        bt_values = sample_values.get(feat_name, [])
        
        if not bt_values or bq_std == 0:
            print(f'{feat_name:<20} {"N/A":>10} {"N/A":>10} {"N/A":>10} {"N/A":>10} {"N/A":>10} {"SKIP":<8}')
            continue
        
        bt_mean = np.mean(bt_values)
        bt_std = np.std(bt_values)
        shift = abs(bt_mean - bq_mean) / bq_std  # Z-score of mean shift
        
        status = 'OK' if shift < 2.0 else 'DRIFT'
        if shift >= 2.0:
            drift_detected = True
        
        print(f'{feat_name:<20} {bq_mean:>10.3f} {bq_std:>10.3f} {bt_mean:>10.3f} {bt_std:>10.3f} {shift:>10.2f} {status:<8}')

    print(f'\n{"⚠ Drift detected — consider retraining" if drift_detected else "All features within normal range"}')
````

### Automated Drift Detection

In production, run drift checks on a schedule using [Cloud Scheduler](https://cloud.google.com/scheduler/docs/overview) + [Cloud Functions](https://cloud.google.com/functions/docs/concepts/overview):

1. **Cloud Scheduler** triggers a Cloud Function every hour
2. **Cloud Function** reads a sample from Bigtable, computes statistics, compares to stored baselines
3. If drift exceeds threshold → write a [custom metric](https://cloud.google.com/monitoring/custom-metrics/creating-metrics) to Cloud Monitoring
4. **Cloud Monitoring alert** fires when the custom metric crosses the threshold


### BigQuery ML Data Validation Functions

The manual approach above works for understanding drift detection. For production, BigQuery ML provides purpose-built functions that handle the statistics, comparison, and anomaly detection automatically:

| Function | Purpose | Use Case |
|----------|---------|----------|
| [`ML.DESCRIBE_DATA`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-describe-data) | Compute descriptive statistics for training or serving data | Baseline computation |
| [`ML.VALIDATE_DATA_SKEW`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-validate-data-skew) | Compare serving data statistics against training data statistics | Training-serving skew detection |
| [`ML.VALIDATE_DATA_DRIFT`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-validate-data-drift) | Compare two sets of serving data to identify anomalous differences | Drift over time |
| [`ML.TFDV_DESCRIBE`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-tfdv-describe) | Fine-grained statistics (same as TensorFlow `tfdv.generate_statistics_from_csv`) | Detailed statistical profiling |
| [`ML.TFDV_VALIDATE`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-tfdv-validate) | Compare training vs serving or serving vs serving (same as TensorFlow `validate_statistics`) | Schema and distribution validation |

These functions integrate naturally with a Bigtable feature store workflow: the offline store (BigQuery) validates data quality, and any skew or drift it detects tells you whether the online store (Bigtable) is serving stale or shifted feature values.


````python
# BigQuery ML data validation — production-grade drift detection
#
# ML.DESCRIBE_DATA computes baseline statistics from training data.
# ML.VALIDATE_DATA_SKEW compares serving data against those baselines.
# ML.VALIDATE_DATA_DRIFT compares two sets of serving data over time.

# Step 1: Describe training data (compute baseline statistics)
describe_query = f'''
SELECT * FROM ML.DESCRIBE_DATA(
    (SELECT
        feature_float_1, feature_float_2, feature_float_3,
        feature_int_1, feature_int_2
    FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
    LIMIT 50000),
    STRUCT(0.8 AS categorical_default_threshold)
)
'''

try:
    describe_df = bq.query(describe_query).to_dataframe()
    print('ML.DESCRIBE_DATA — Training data statistics:')
    print(describe_df.to_string())
except Exception as e:
    print(f'ML.DESCRIBE_DATA: {e}')
    print('This function requires BigQuery ML to be enabled.')

# Step 2: Validate data skew (training vs serving)
# In production, serving data would come from a BigQuery external table
# pointing at Bigtable, or from a periodic export of served features.
skew_query = f'''
SELECT * FROM ML.VALIDATE_DATA_SKEW(
    -- Training data (baseline)
    (SELECT feature_float_1, feature_float_2, feature_float_3,
            feature_int_1, feature_int_2
     FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
     LIMIT 50000),
    -- Serving data (current — simulated with a different sample)
    (SELECT feature_float_1, feature_float_2, feature_float_3,
            feature_int_1, feature_int_2
     FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
     ORDER BY RAND()
     LIMIT 10000),
    STRUCT(0.01 AS categorical_default_threshold)
)
'''

try:
    skew_df = bq.query(skew_query).to_dataframe()
    print('\nML.VALIDATE_DATA_SKEW — Training vs Serving comparison:')
    print(skew_df.to_string())
except Exception as e:
    print(f'\nML.VALIDATE_DATA_SKEW: {e}')

# Step 3: Validate data drift (serving at time T1 vs serving at time T2)
drift_query = f'''
SELECT * FROM ML.VALIDATE_DATA_DRIFT(
    -- Serving data at time T1 (earlier window)
    (SELECT feature_float_1, feature_float_2, feature_float_3
     FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
     WHERE entity_group IN ('A', 'B', 'C')),
    -- Serving data at time T2 (later window)
    (SELECT feature_float_1, feature_float_2, feature_float_3
     FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
     WHERE entity_group IN ('D', 'E', 'F')),
    STRUCT(0.01 AS categorical_default_threshold)
)
'''

try:
    drift_df = bq.query(drift_query).to_dataframe()
    print('\nML.VALIDATE_DATA_DRIFT — Serving data over time:')
    print(drift_df.to_string())
except Exception as e:
    print(f'\nML.VALIDATE_DATA_DRIFT: {e}')
````

### Putting It Together: Bigtable + BigQuery ML Monitoring

The two approaches complement each other:

| Layer | Tool | What It Catches | When to Use |
|-------|------|-----------------|-------------|
| **Online (Bigtable)** | Manual sampling + Z-scores | Real-time serving drift — are features in Bigtable still matching training? | Continuous monitoring, alerts |
| **Offline (BigQuery)** | `ML.VALIDATE_DATA_SKEW` | Training-serving skew — did the export pipeline change the data? | After each export or sync |
| **Offline (BigQuery)** | `ML.VALIDATE_DATA_DRIFT` | Temporal drift — are feature distributions shifting over time? | Periodic (daily/weekly) checks |
| **Offline (BigQuery)** | `ML.DESCRIBE_DATA` | Baseline profiling — what does the training data look like? | At training time, stored as reference |

A typical production workflow:
1. At training time, run `ML.DESCRIBE_DATA` and save the statistics
2. After each EXPORT DATA sync, run `ML.VALIDATE_DATA_SKEW` against training baselines
3. Weekly, run `ML.VALIDATE_DATA_DRIFT` comparing this week's serving data to last week's
4. Continuously, sample Bigtable features and compare to baselines (the manual approach above)
5. Alert on any anomalies — retrain if drift is confirmed


## 6. App Profiles

[App profiles](https://cloud.google.com/bigtable/docs/app-profiles) control how Bigtable routes and prioritizes requests. Different traffic patterns need different routing:

| Profile | Routing | Priority | Use case |
|---------|---------|----------|----------|
| `default` | Multi-cluster (any) | HIGH | General access |
| `serving-profile` | Multi-cluster | HIGH | Real-time ML serving |
| `bq-export` | Single-cluster | LOW | BigQuery EXPORT DATA |
| `batch-reads` | Multi-cluster | MEDIUM | Batch feature extraction |

### Routing Types

- **Single-cluster routing** sends all requests to one specific cluster. Required for `EXPORT DATA` (which needs consistent write ordering) and strongly-consistent reads.
- **Multi-cluster routing** distributes requests across all clusters for high availability. Reads may be eventually consistent but benefit from geographic proximity and failover.

### Priority Levels

- **PRIORITY_HIGH** — latency-sensitive serving traffic gets first access to resources
- **PRIORITY_MEDIUM** — batch reads that can tolerate some queuing
- **PRIORITY_LOW** — background jobs (exports, backfills) that should yield to serving

Using different priorities prevents a large EXPORT DATA job from starving real-time serving reads.

````python
# List existing app profiles
result = subprocess.run(
    ['gcloud', 'bigtable', 'app-profiles', 'list',
     '--instance', BT_INSTANCE,
     '--project', PROJECT_ID,
     '--format', 'table(name.basename(), routingPolicy, priority, description)'],
    capture_output=True, text=True,
)

if result.returncode == 0:
    print('Existing app profiles:')
    print(result.stdout)
else:
    print(f'Could not list profiles: {result.stderr.strip()}')
````

````python
# Create a serving profile with HIGH priority
# Get cluster ID first
cluster_result = subprocess.run(
    ['gcloud', 'bigtable', 'clusters', 'list',
     '--instances', BT_INSTANCE,
     '--project', PROJECT_ID,
     '--format', 'value(name.basename())'],
    capture_output=True, text=True,
)
cluster_id = cluster_result.stdout.strip().split('\n')[0] if cluster_result.returncode == 0 else f'{BT_INSTANCE}-c1'
print(f'Cluster ID: {cluster_id}')

# Create the serving profile
result = subprocess.run(
    ['gcloud', 'bigtable', 'app-profiles', 'create', 'serving-profile',
     '--instance', BT_INSTANCE,
     '--project', PROJECT_ID,
     '--route-to', cluster_id,
     '--priority', 'PRIORITY_HIGH',
     '--description', 'Real-time ML serving - HIGH priority',
     '--force'],
    capture_output=True, text=True,
)

if result.returncode == 0:
    print(f'Created app profile: serving-profile')
elif 'already exists' in result.stderr.lower():
    print(f'App profile already exists: serving-profile')
else:
    print(f'Note: {result.stderr.strip()}')
````

````python
# Create a batch-reads profile with MEDIUM priority
result = subprocess.run(
    ['gcloud', 'bigtable', 'app-profiles', 'create', 'batch-reads',
     '--instance', BT_INSTANCE,
     '--project', PROJECT_ID,
     '--route-to', cluster_id,
     '--priority', 'PRIORITY_MEDIUM',
     '--description', 'Batch feature extraction - MEDIUM priority',
     '--force'],
    capture_output=True, text=True,
)

if result.returncode == 0:
    print(f'Created app profile: batch-reads')
elif 'already exists' in result.stderr.lower():
    print(f'App profile already exists: batch-reads')
else:
    print(f'Note: {result.stderr.strip()}')
````

````python
# List all profiles after creation
result = subprocess.run(
    ['gcloud', 'bigtable', 'app-profiles', 'list',
     '--instance', BT_INSTANCE,
     '--project', PROJECT_ID,
     '--format', 'table(name.basename(), routingPolicy, priority, description)'],
    capture_output=True, text=True,
)

if result.returncode == 0:
    print('All app profiles:')
    print(result.stdout)
````

### Using App Profiles in Code

To use a specific app profile, pass it when opening the table:

```python
# Serving reads use HIGH priority
serving_table = instance.table(BT_TABLE, app_profile_id='serving-profile')
row = serving_table.read_row(user_id)

# Batch reads use MEDIUM priority
batch_table = instance.table(BT_TABLE, app_profile_id='batch-reads')
rows = batch_table.read_rows(start_key=start, end_key=end)

# EXPORT DATA uses LOW priority (in the Bigtable URI)
# https://bigtable.googleapis.com/projects/.../appProfiles/bq-export/tables/...
```

The serving pipeline should always use the HIGH-priority profile. Background jobs (backfills, exports, analytics) use MEDIUM or LOW so they never starve serving traffic.

## 7. Cost Optimization

Bigtable pricing has three components: **nodes** (compute), **storage** (SSD or HDD), and **network** (egress). For a feature store, nodes dominate the cost.

### Node Types and Storage

| Storage type | Read latency | Cost (per GB/month) | Use case |
|-------------|-------------|--------------------|---------|
| **SSD** | Single-digit ms | ~$0.17 | Feature stores, serving (required) |
| **HDD** | 200ms+ | ~$0.026 | Archival, analytics, cold storage |

For ML feature stores, **SSD is required**. The entire value proposition is single-digit millisecond reads — HDD's 200ms+ latency defeats the purpose.

### Node Scaling

| Instance type | Min nodes | SLA | Cost (per node/hr) | Use case |
|--------------|-----------|-----|--------------------|---------|
| Development | 1 (fixed) | None | ~$0.65 | Labs, prototyping |
| Production | 3+ | 99.99% | ~$0.65 | Production workloads |
| Production + autoscaling | 3-30 (auto) | 99.99% | Variable | Variable traffic |

### Cost Estimate for Our Feature Store

Let's estimate the monthly cost for our 130K-entity feature store.

````python
# Cost estimation for our feature store
NUM_ENTITIES = 130_000
AVG_ROW_SIZE_BYTES = 150  # 10 features, pipe-delimited, ~15 bytes each

# Storage
total_storage_gb = (NUM_ENTITIES * AVG_ROW_SIZE_BYTES) / (1024 ** 3)
ssd_cost_per_gb = 0.17  # per GB per month
storage_cost = total_storage_gb * ssd_cost_per_gb

# Nodes
node_cost_per_hr = 0.65
hours_per_month = 730

# Scenarios
scenarios = [
    ('Development (1 node, no SLA)', 1),
    ('Production minimum (3 nodes)', 3),
    ('Production with headroom (5 nodes)', 5),
]

print(f'Feature store cost estimate:')
print(f'  Entities:     {NUM_ENTITIES:,}')
print(f'  Row size:     ~{AVG_ROW_SIZE_BYTES} bytes')
print(f'  Total data:   {total_storage_gb * 1024:.1f} MB ({total_storage_gb:.4f} GB)')
print(f'  Storage cost: ${storage_cost:.2f}/month (SSD)')
print()

print(f'{"Scenario":<45s} {"Nodes":>5s} {"Node Cost":>12s} {"Storage":>10s} {"Total":>10s}')
print('-' * 85)
for name, nodes in scenarios:
    node_cost = nodes * node_cost_per_hr * hours_per_month
    total = node_cost + storage_cost
    print(f'{name:<45s} {nodes:>5d} ${node_cost:>10,.2f} ${storage_cost:>8.2f} ${total:>8,.2f}')

print(f'\nNote: Network egress adds ~$0.01-0.12/GB depending on destination.')
print(f'For serving within the same region, egress is free.')
````

### Committed Use Discounts

For predictable workloads, [Committed Use Discounts (CUDs)](https://cloud.google.com/bigtable/pricing#committed-use-discounts) reduce node costs significantly:

| Commitment | Discount | Effective node cost/hr |
|-----------|----------|----------------------|
| None (on-demand) | 0% | $0.65 |
| 1-year | ~25% | ~$0.49 |
| 3-year | ~50% | ~$0.33 |

CUDs apply to node-hours. Storage and network are not discountable. For a production feature store that runs 24/7, a 1-year commitment is almost always worth it.

### Autoscaling

Bigtable autoscaling adjusts node count based on CPU utilization:

```bash
gcloud bigtable clusters update CLUSTER_ID \
    --instance=INSTANCE_ID \
    --autoscaling-min-nodes=3 \
    --autoscaling-max-nodes=10 \
    --autoscaling-cpu-target=60
```

For feature stores with variable traffic (e.g., e-commerce with peak hours), autoscaling prevents over-provisioning during off-peak while ensuring capacity during peaks.
