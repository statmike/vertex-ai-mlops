# Bigtable Feature Store — History & Time-Travel

ML models need point-in-time correct features: "what did this entity's features look like at 2pm last Tuesday?" Getting this wrong causes training-serving skew. This notebook compares two approaches to storing feature history in Bigtable and shows how to extract point-in-time features for training.

## Section 1: Why History Matters

A feature store that only keeps the latest value works for serving. But training requires **point-in-time correctness** — the ability to reconstruct what features looked like at any historical moment. This is related to BigQuery's [time travel](https://cloud.google.com/bigquery/docs/time-travel) concept, but applied at the feature level.

### Training-Serving Skew

Consider a fraud model. At training time, you join features to labeled transactions using the transaction timestamp. If you accidentally use *current* feature values instead of *historical* values, the model learns patterns that won't exist at serving time. This is **training-serving skew** — the most common silent failure in production ML.

```
Transaction at 2024-06-15 14:00:00 UTC:

  WRONG: Use today's feature values (spending_30d_avg = $142.50)
         The model sees the future — values influenced by events after the transaction.

  RIGHT: Use feature values as-of 2024-06-15 14:00:00 UTC (spending_30d_avg = $87.20)
         The model sees exactly what was known at decision time.
```

### Point-in-Time Correctness

For a transaction at time T, you need the feature values that were **known at time T**, not values computed later. This matters for:

| Use Case | Why PIT Matters |
|----------|----------------|
| **Training** | Features must reflect what was known when the label was generated |
| **Backtesting** | Simulating model performance on historical data requires historical features |
| **Auditing** | Regulators may ask "what did the model see when it made this decision?" |
| **Debugging** | Reproducing a bad prediction requires the exact feature state at that time |

### Two Storage Strategies

Bigtable offers two natural ways to store feature history, each with different trade-offs:

- **Design A: History in the Row Key** — encode the timestamp in the row key so each snapshot is a separate row
- **Design B: Cell Versioning** — use Bigtable's native cell versioning to store multiple values per column

We'll implement both, compare them, and show when to use which.

---
## Section 2: Design Choice A — History in the Row Key

In this design, each feature snapshot gets its own row. The row key encodes both the entity identity and the timestamp:

```
Row key format: {entity_group}#{entity_id}#{reverse_timestamp}

Example rows:
  A#000001#9999999999999 → oldest snapshot
  A#000001#9999999999500 → newer snapshot
  A#000001#9999999999000 → newest snapshot
```

**Reverse timestamp:** We use `MAX_TIMESTAMP - event_timestamp` so that the newest entries sort first in Bigtable's lexicographic ordering. This makes "get the latest N snapshots" a simple prefix scan with a row limit.

### Create the Table

````python
# Create a dedicated table for Design A: history in the key
table_history_key = instance.table(BT_TABLE_HISTORY_KEY)

if not table_history_key.exists():
    # No automatic GC — history retention is managed manually
    max_age_rule = cf.MaxAgeGCRule(datetime.timedelta(days=365))
    table_history_key.create(column_families={BT_COLUMN_FAMILY: max_age_rule})
    print(f'Created table: {BT_TABLE_HISTORY_KEY}')
    print(f'  Column family: {BT_COLUMN_FAMILY}')
    print(f'  GC policy: MaxAge(365 days)')
else:
    print(f'Table exists: {BT_TABLE_HISTORY_KEY}')
    existing = list(table_history_key.list_column_families().keys())
    print(f'  Column families: {existing}')
````

### Sample Data from BigQuery

We'll pull events for 10 entities from `sparse_events`. Each entity has multiple events over time — these become the historical snapshots we write to Bigtable.

````python
# Select 10 entities with multiple events
sample_df = bq.query(f'''
    WITH entity_counts AS (
        SELECT entity_group, entity_id, COUNT(*) AS event_count
        FROM `{DATASET_REF}.sparse_events`
        GROUP BY entity_group, entity_id
        HAVING COUNT(*) >= 5
        ORDER BY FARM_FINGERPRINT(CONCAT(entity_group, entity_id))
        LIMIT 10
    )
    SELECT
        e.entity_group,
        e.entity_id,
        e.event_timestamp,
        e.feature_1,
        e.feature_2,
        e.feature_3
    FROM `{DATASET_REF}.sparse_events` e
    JOIN entity_counts c
        ON e.entity_group = c.entity_group AND e.entity_id = c.entity_id
    ORDER BY e.entity_group, e.entity_id, e.event_timestamp
''').to_dataframe()

print(f'Sample data: {len(sample_df)} events across {sample_df[["entity_group", "entity_id"]].drop_duplicates().shape[0]} entities')
print(f'\nEntities:')
entity_summary = sample_df.groupby(['entity_group', 'entity_id']).agg(
    events=('event_timestamp', 'count'),
    earliest=('event_timestamp', 'min'),
    latest=('event_timestamp', 'max'),
).reset_index()
print(entity_summary.to_string(index=False))
````

### Write Historical Snapshots with Compound Keys

Each event becomes a separate row in Bigtable. The row key encodes entity identity and a reverse timestamp so newest entries sort first.

````python
# Maximum timestamp for reverse key computation
MAX_TIMESTAMP = 9999999999999  # 13-digit max (milliseconds)

def make_history_key(entity_group, entity_id, event_timestamp):
    """Build a compound row key with reverse timestamp.
    
    Format: {entity_group}#{entity_id}#{reverse_timestamp}
    Reverse timestamp ensures newest entries sort first.
    """
    if isinstance(event_timestamp, pd.Timestamp):
        ts_millis = int(event_timestamp.timestamp() * 1000)
    else:
        ts_millis = int(event_timestamp * 1000)
    reverse_ts = MAX_TIMESTAMP - ts_millis
    return f'{entity_group}#{entity_id}#{reverse_ts:013d}'


def extract_timestamp_from_key(row_key):
    """Extract the original timestamp from a compound row key."""
    parts = row_key.split('#')
    reverse_ts = int(parts[2])
    ts_millis = MAX_TIMESTAMP - reverse_ts
    return datetime.datetime.fromtimestamp(ts_millis / 1000, tz=datetime.timezone.utc)


# Write sample data to Bigtable
rows_written = 0
rows = []

for _, event in sample_df.iterrows():
    row_key = make_history_key(
        event['entity_group'],
        event['entity_id'],
        event['event_timestamp'],
    )
    
    row = table_history_key.direct_row(row_key)
    
    # Write each feature as a separate column qualifier
    for col in ['feature_1', 'feature_2', 'feature_3']:
        value = event[col]
        if pd.notna(value):
            row.set_cell(
                BT_COLUMN_FAMILY,
                col,
                str(value).encode('utf-8'),
            )
    
    rows.append(row)
    rows_written += 1

# Batch mutate for efficiency
responses = table_history_key.mutate_rows(rows)
errors = [r for r in responses if r.code != 0]

print(f'Wrote {rows_written} rows to {BT_TABLE_HISTORY_KEY}')
if errors:
    print(f'  Errors: {len(errors)}')
else:
    print(f'  All writes successful')

# Show a few example keys
print(f'\nExample row keys:')
for _, event in sample_df.head(3).iterrows():
    key = make_history_key(event['entity_group'], event['entity_id'], event['event_timestamp'])
    print(f'  {key}  →  {event["event_timestamp"]}')
````

### Range Scan: Read All History for One Entity

Since all rows for an entity share the same prefix (`{entity_group}#{entity_id}#`), a [prefix scan](https://cloud.google.com/bigtable/docs/reads) retrieves the full history. The reverse timestamp ordering means rows come back newest-first.

````python
# Pick the first entity from our sample
test_entity_group = sample_df.iloc[0]['entity_group']
test_entity_id = sample_df.iloc[0]['entity_id']
prefix = f'{test_entity_group}#{test_entity_id}#'

print(f'Reading all history for entity: {test_entity_group}#{test_entity_id}')
print(f'Row key prefix: {prefix}')
print()

# Prefix scan
row_set = RowSet()
row_set.add_row_range(RowRange(start_key=prefix.encode('utf-8'),
                               end_key=(prefix[:-1] + chr(ord(prefix[-1]) + 1)).encode('utf-8')))

partial_rows = table_history_key.read_rows(
    row_set=row_set,
    filter_=row_filters.CellsColumnLimitFilter(1),
)
partial_rows.consume_all()

print(f'Found {len(partial_rows.rows)} historical snapshots:\n')

history_records = []
for row_key_bytes, row_data in sorted(partial_rows.rows.items()):
    row_key = row_key_bytes.decode('utf-8')
    original_ts = extract_timestamp_from_key(row_key)
    
    record = {'row_key': row_key, 'timestamp': original_ts}
    for col_bytes, cells in row_data.cells.get(BT_COLUMN_FAMILY, {}).items():
        col_name = col_bytes.decode('utf-8')
        record[col_name] = cells[0].value.decode('utf-8')
    history_records.append(record)

history_df = pd.DataFrame(history_records)
print(history_df.to_string(index=False))
````

### Point-in-Time Read: Features As-Of a Specific Time

To get features as-of time T, we need the **first row** that sorts after the reverse timestamp of T. Since newer timestamps have smaller reverse timestamps (sort earlier), we scan starting from the reverse timestamp of T and take the first result — that's the most recent snapshot at or before time T.

````python
def read_features_at_time_key_design(table, entity_group, entity_id, as_of_timestamp):
    """Read features as-of a specific time using Design A (history in key).
    
    The reverse timestamp means we need the first row whose reverse_ts
    is >= the reverse of our target time (i.e., original timestamp <= target).
    """
    if isinstance(as_of_timestamp, str):
        as_of_timestamp = pd.Timestamp(as_of_timestamp)
    
    ts_millis = int(as_of_timestamp.timestamp() * 1000)
    reverse_ts = MAX_TIMESTAMP - ts_millis
    
    # Start key: entity prefix + reverse of target time
    start_key = f'{entity_group}#{entity_id}#{reverse_ts:013d}'
    # End key: everything up to next entity
    end_prefix = f'{entity_group}#{entity_id}$'  # '$' > '#' in ASCII
    
    row_set = RowSet()
    row_set.add_row_range(RowRange(
        start_key=start_key.encode('utf-8'),
        end_key=end_prefix.encode('utf-8'),
    ))
    
    partial_rows = table.read_rows(
        row_set=row_set,
        filter_=row_filters.CellsColumnLimitFilter(1),
        limit=1,  # We only need the first (most recent at-or-before T)
    )
    partial_rows.consume_all()
    
    if not partial_rows.rows:
        return None
    
    row_key_bytes = list(partial_rows.rows.keys())[0]
    row_data = partial_rows.rows[row_key_bytes]
    row_key = row_key_bytes.decode('utf-8')
    original_ts = extract_timestamp_from_key(row_key)
    
    features = {'_timestamp': original_ts, '_row_key': row_key}
    for col_bytes, cells in row_data.cells.get(BT_COLUMN_FAMILY, {}).items():
        features[col_bytes.decode('utf-8')] = cells[0].value.decode('utf-8')
    
    return features


# Demonstrate: pick a time between two known events
entity_events = sample_df[
    (sample_df['entity_group'] == test_entity_group) &
    (sample_df['entity_id'] == test_entity_id)
].sort_values('event_timestamp')

if len(entity_events) >= 2:
    # Choose a time between the 2nd and 3rd events
    idx = min(1, len(entity_events) - 2)
    t_before = entity_events.iloc[idx]['event_timestamp']
    t_after = entity_events.iloc[idx + 1]['event_timestamp']
    
    # Query at midpoint
    midpoint = t_before + (t_after - t_before) / 2
    
    print(f'Entity: {test_entity_group}#{test_entity_id}')
    print(f'\nKnown events:')
    print(f'  Event {idx}:   {t_before}')
    print(f'  Midpoint:  {midpoint}  <-- querying at this time')
    print(f'  Event {idx+1}:   {t_after}')
    
    result = read_features_at_time_key_design(
        table_history_key, test_entity_group, test_entity_id, midpoint
    )
    
    if result:
        print(f'\nFeatures as-of {midpoint}:')
        for k, v in result.items():
            print(f'  {k}: {v}')
        print(f'\nCorrectness: returned snapshot from {result["_timestamp"]} (should be <= {midpoint})')
    else:
        print('\nNo features found at that time.')
else:
    print(f'Entity has only {len(entity_events)} events — need at least 2 for demonstration.')
````

### Design A: Pros and Cons

| Aspect | Assessment |
|--------|------------|
| **Range scans** | Natural — prefix scan retrieves all history for an entity in timestamp order |
| **Training data extraction** | Efficient — scan all entities in a range and filter by time |
| **Latest value** | Prefix scan + limit 1 (slightly more expensive than a direct read) |
| **Garbage collection** | Manual — you must delete old rows yourself or rely on MaxAge GC on the column family |
| **Hotspotting** | Risk if many entities write at similar times (sequential reverse timestamps) |
| **Storage** | More rows = more row key overhead |
| **Serving key compatibility** | Different key format than a simple `entity_id` serving table |

---
## Section 3: Design Choice B — Cell Versioning

Bigtable natively supports multiple versions of each cell, each tagged with a timestamp. Instead of encoding history in the row key, we store all versions in the same row:

```
Row key: {entity_group}#{entity_id}

Cell versions for column "feature_1":
  2024-06-15 14:00:00  →  42.5
  2024-06-14 10:30:00  →  38.2
  2024-06-13 09:15:00  →  35.0
```

This is Bigtable's native time-series capability. Each `set_cell()` call can include a `timestamp` parameter, and Bigtable stores the value as a separate version.

### Create the Table with MaxVersionsGCRule

````python
# Create a dedicated table for Design B: cell versioning
table_history_cell = instance.table(BT_TABLE_HISTORY_CELL)

if not table_history_cell.exists():
    # Automatic GC: keep at most 10 versions per cell
    max_versions_rule = cf.MaxVersionsGCRule(10)
    table_history_cell.create(column_families={BT_COLUMN_FAMILY: max_versions_rule})
    print(f'Created table: {BT_TABLE_HISTORY_CELL}')
    print(f'  Column family: {BT_COLUMN_FAMILY}')
    print(f'  GC policy: MaxVersions(10)')
else:
    print(f'Table exists: {BT_TABLE_HISTORY_CELL}')
    families = table_history_cell.list_column_families()
    for name, family in families.items():
        print(f'  Family: {name}, GC rule: {family.gc_rule}')
````

### Write Historical Snapshots as Cell Versions

Same 10 entities, but now each event is written as a new cell version on the same row key. The `timestamp` parameter in `set_cell()` controls which version is created.

````python
# Write sample data using cell versioning
rows_written = 0
rows = []

for _, event in sample_df.iterrows():
    # Simple key: entity_group#entity_id (no timestamp)
    row_key = f'{event["entity_group"]}#{event["entity_id"]}'
    
    row = table_history_cell.direct_row(row_key)
    
    # Convert event timestamp to Bigtable timestamp (datetime object)
    event_ts = event['event_timestamp']
    if isinstance(event_ts, pd.Timestamp):
        bt_timestamp = event_ts.to_pydatetime()
    else:
        bt_timestamp = datetime.datetime.fromtimestamp(
            event_ts, tz=datetime.timezone.utc
        )
    
    # Write each feature with the event timestamp as the cell version
    for col in ['feature_1', 'feature_2', 'feature_3']:
        value = event[col]
        if pd.notna(value):
            row.set_cell(
                BT_COLUMN_FAMILY,
                col,
                str(value).encode('utf-8'),
                timestamp=bt_timestamp,
            )
    
    rows.append(row)
    rows_written += 1

# Batch mutate
responses = table_history_cell.mutate_rows(rows)
errors = [r for r in responses if r.code != 0]

print(f'Wrote {rows_written} cell versions to {BT_TABLE_HISTORY_CELL}')
if errors:
    print(f'  Errors: {len(errors)}')
else:
    print(f'  All writes successful')

# Show unique row keys (should be fewer than total writes)
unique_keys = sample_df.apply(
    lambda r: f'{r["entity_group"]}#{r["entity_id"]}', axis=1
).nunique()
print(f'\n  Total writes: {rows_written}')
print(f'  Unique row keys: {unique_keys}')
print(f'  Average versions per row: {rows_written / unique_keys:.1f}')
````

### Read Latest: CellsColumnLimitFilter

For serving, you typically want only the latest version of each feature. `CellsColumnLimitFilter(1)` returns just the newest cell version per column.

````python
# Read only the latest version for our test entity
test_key = f'{test_entity_group}#{test_entity_id}'

row = table_history_cell.read_row(
    test_key.encode('utf-8'),
    filter_=row_filters.CellsColumnLimitFilter(1),
)

if row:
    print(f'Latest features for {test_key}:\n')
    for col_bytes, cells in sorted(row.cells.get(BT_COLUMN_FAMILY, {}).items()):
        col_name = col_bytes.decode('utf-8')
        cell = cells[0]
        print(f'  {col_name}: {cell.value.decode("utf-8")}  (version: {cell.timestamp})')
else:
    print(f'No row found for {test_key}')
````

### Read at Time T: TimestampRangeFilter

To get features as-of time T, use a [`TimestampRangeFilter`](https://cloud.google.com/bigtable/docs/using-filters) that restricts cell versions to those at or before T. Combined with `CellsColumnLimitFilter(1)`, this returns the most recent version at or before the target time.

````python
def read_features_at_time_cell_design(table, entity_group, entity_id, as_of_timestamp):
    """Read features as-of a specific time using Design B (cell versioning).
    
    Uses TimestampRangeFilter to restrict to versions <= as_of_timestamp,
    then CellsColumnLimitFilter(1) to get only the most recent such version.
    """
    if isinstance(as_of_timestamp, str):
        as_of_timestamp = pd.Timestamp(as_of_timestamp)
    
    if isinstance(as_of_timestamp, pd.Timestamp):
        as_of_dt = as_of_timestamp.to_pydatetime()
    else:
        as_of_dt = as_of_timestamp
    
    # Chain: timestamp range (end is exclusive) + limit 1 per column
    time_filter = row_filters.TimestampRangeFilter(
        start=None,
        end=as_of_dt,
    )
    limit_filter = row_filters.CellsColumnLimitFilter(1)
    chained = row_filters.RowFilterChain(filters=[time_filter, limit_filter])
    
    row_key = f'{entity_group}#{entity_id}'
    row = table.read_row(row_key.encode('utf-8'), filter_=chained)
    
    if not row:
        return None
    
    features = {}
    for col_bytes, cells in row.cells.get(BT_COLUMN_FAMILY, {}).items():
        col_name = col_bytes.decode('utf-8')
        cell = cells[0]
        features[col_name] = cell.value.decode('utf-8')
        features[f'{col_name}_timestamp'] = cell.timestamp
    
    return features


# Demonstrate: same midpoint query as Design A
if len(entity_events) >= 2:
    idx = min(1, len(entity_events) - 2)
    t_before = entity_events.iloc[idx]['event_timestamp']
    t_after = entity_events.iloc[idx + 1]['event_timestamp']
    midpoint = t_before + (t_after - t_before) / 2
    
    print(f'Entity: {test_entity_group}#{test_entity_id}')
    print(f'Querying as-of: {midpoint}')
    print()
    
    result = read_features_at_time_cell_design(
        table_history_cell, test_entity_group, test_entity_id, midpoint
    )
    
    if result:
        print(f'Features as-of {midpoint}:')
        for k, v in result.items():
            if not k.endswith('_timestamp'):
                ts_key = f'{k}_timestamp'
                ts_val = result.get(ts_key, 'N/A')
                print(f'  {k}: {v}  (version: {ts_val})')
    else:
        print('No features found at that time.')
````

### Read All Versions

To see the full history stored in a row, read without the column limit filter (or use `CellsColumnLimitFilter(N)` for the N most recent).

````python
# Read all versions for one entity
test_key = f'{test_entity_group}#{test_entity_id}'

# No CellsColumnLimitFilter — get all versions
row = table_history_cell.read_row(test_key.encode('utf-8'))

if row:
    print(f'All cell versions for {test_key}:\n')
    for col_bytes, cells in sorted(row.cells.get(BT_COLUMN_FAMILY, {}).items()):
        col_name = col_bytes.decode('utf-8')
        print(f'  {col_name} ({len(cells)} versions):')
        for cell in cells:
            print(f'    {cell.timestamp}  →  {cell.value.decode("utf-8")}')
        print()
else:
    print(f'No row found for {test_key}')
````

### Design B: Pros and Cons

| Aspect | Assessment |
|--------|------------|
| **Serving key** | Same key format as a serving-only table — no key translation needed |
| **Latest read** | `CellsColumnLimitFilter(1)` — fast, simple |
| **Automatic GC** | `MaxVersionsGCRule` handles cleanup automatically |
| **Fewer rows** | All history for an entity lives in one row |
| **Historical read** | Requires filters (`TimestampRangeFilter`) — slightly more complex |
| **Row size** | Many versions = larger rows; can hit row size limits for very high-frequency entities |
| **Range scans** | Cannot scan across entities by time range without reading all rows |

## Section 4: Point-in-Time Joins with Bigtable Filters

We'll implement this for both designs and compare with BigQuery's `ML.FEATURES_AT_TIME`.

### Compare with BigQuery: ML.FEATURES_AT_TIME

BigQuery provides [`ML.FEATURES_AT_TIME`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-feature-time) for point-in-time feature retrieval directly in SQL. This is the batch analytics approach — no Bigtable needed, but with higher latency.

````python
# Pick one entity and timestamp for demonstration
demo_entity = pit_df.iloc[0]
demo_ts = demo_entity['query_timestamp'].strftime('%Y-%m-%d %H:%M:%S UTC')
demo_group = demo_entity['entity_group']
demo_id = demo_entity['entity_id']

print(f'ML.FEATURES_AT_TIME query for entity {demo_group}/{demo_id} at {demo_ts}\n')

bq_pit_sql = f"""
SELECT * FROM ML.FEATURES_AT_TIME(
    TABLE `{BQ_PROJECT}.{BQ_DATASET}.sparse_events`,
    time => '{demo_ts}',
    num_rows => 1,
    ignore_feature_nulls => TRUE
)
WHERE entity_group = '{demo_group}' AND entity_id = '{demo_id}'
"""

start_time = time.perf_counter()
try:
    bq_result = bq.query(bq_pit_sql).to_dataframe()
    bq_pit_time = time.perf_counter() - start_time
    
    print(f'BQ result ({bq_pit_time:.2f}s):')
    print(bq_result.to_string(index=False))
except Exception as e:
    bq_pit_time = time.perf_counter() - start_time
    print(f'BQ query failed ({bq_pit_time:.2f}s): {e}')
    print('\nNote: ML.FEATURES_AT_TIME requires the source table to have')
    print('appropriate columns (entity key + feature columns + timestamp).')
    print('If sparse_events uses a different schema, the query needs adjustment.')
````

### Pattern 1: BigQuery Bulk Extraction

Use BigQuery's [`ML.ENTITY_FEATURES_AT_TIME`](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-entity-feature-time) when you have a large entity-time table (your label table) and want features for all of them at once. BigQuery handles the PIT join internally.

````python
# Pattern 1: BQ bulk extraction using ML.ENTITY_FEATURES_AT_TIME
# First, create a temporary entity-time table from our sample

# Build the entity-time pairs as a query
entity_time_sql = f"""
WITH entity_times AS (
    SELECT
        entity_group,
        entity_id,
        event_timestamp AS feature_timestamp
    FROM `{DATASET_REF}.sparse_events`
    WHERE entity_group = '{demo_group}'
    ORDER BY event_timestamp
    LIMIT 20
)
SELECT
    entity_group,
    entity_id,
    feature_timestamp
FROM entity_times
"""

print('Pattern 1: BigQuery ML.ENTITY_FEATURES_AT_TIME')
print('This joins a feature table with an entity-time table to get PIT features.\n')

# Show the SQL pattern (may need schema adjustment for the actual table)
bulk_sql = f"""
SELECT * FROM ML.ENTITY_FEATURES_AT_TIME(
    TABLE `{BQ_PROJECT}.{BQ_DATASET}.sparse_events`,
    ({entity_time_sql}),
    num_rows => 1,
    ignore_feature_nulls => TRUE
)
"""

start_time = time.perf_counter()
try:
    bulk_result = bq.query(bulk_sql).to_dataframe()
    bulk_time = time.perf_counter() - start_time
    print(f'Bulk extraction: {len(bulk_result)} rows in {bulk_time:.2f}s')
    print(f'\nSample:')
    print(bulk_result.head(5).to_string(index=False))
except Exception as e:
    bulk_time = time.perf_counter() - start_time
    print(f'Bulk query result ({bulk_time:.2f}s): {e}')
    print('\nNote: ML.ENTITY_FEATURES_AT_TIME requires specific column naming.')
    print('The entity-time subquery must produce columns matching the feature table keys.')
    print('Adjust column names if the sparse_events schema differs.')
````

### Pattern 2: Bigtable Selective Extraction

When you need features for specific entities at specific times — especially during online inference or when BQ is too slow for interactive use — use Bigtable directly.

````python
# Pattern 2: Bigtable selective extraction
# Useful when you need specific entities at specific times

# Build a list of lookups (simulating a label table)
lookup_requests = pit_df.head(10).to_dict('records')

print('Pattern 2: Bigtable selective extraction')
print(f'Looking up {len(lookup_requests)} entity-time pairs\n')

# Use Design B (cell versioning) for selective lookups
start_time = time.perf_counter()
selective_results = []

for req in lookup_requests:
    features = read_features_at_time_cell_design(
        table_history_cell,
        req['entity_group'],
        req['entity_id'],
        req['query_timestamp'],
    )
    if features:
        features['entity_key'] = f"{req['entity_group']}#{req['entity_id']}"
        features['query_ts'] = req['query_timestamp']
        selective_results.append(features)

selective_time = time.perf_counter() - start_time

print(f'Retrieved {len(selective_results)} results in {selective_time:.3f}s')
print(f'  Avg: {selective_time / len(lookup_requests) * 1000:.1f}ms per lookup')

if selective_results:
    sel_df = pd.DataFrame(selective_results)
    show_cols = [c for c in sel_df.columns if c in 
                 ['entity_key', 'query_ts', 'feature_1', 'feature_2', 'feature_3']]
    print(f'\nResults:')
    print(sel_df[show_cols].to_string(index=False))
````

## Section 6: Garbage Collection Policies

Bigtable's [garbage collection rules](https://cloud.google.com/bigtable/docs/garbage-collection) automatically delete old cell versions. This is how you control the size of your feature history without manual cleanup.

### Available GC Rules

| Rule | What It Does | Example |
|------|-------------|--------|
| `MaxVersionsGCRule(N)` | Keep the N most recent versions | `MaxVersionsGCRule(10)` — keep last 10 snapshots |
| `MaxAgeGCRule(duration)` | Delete versions older than duration | `MaxAgeGCRule(days=90)` — 90-day retention |
| `GCRuleUnion([rules])` | Delete if **any** rule matches (OR) | Delete if older than 90 days OR more than 10 versions |
| `GCRuleIntersection([rules])` | Delete if **all** rules match (AND) | Delete if older than 90 days AND more than 10 versions |

````python
# Demonstrate creating tables with different GC policies

gc_examples = {
    'gc-demo-max-versions': cf.MaxVersionsGCRule(5),
    'gc-demo-max-age': cf.MaxAgeGCRule(datetime.timedelta(days=90)),
    'gc-demo-union': cf.GCRuleUnion([
        cf.MaxVersionsGCRule(10),
        cf.MaxAgeGCRule(datetime.timedelta(days=90)),
    ]),
    'gc-demo-intersection': cf.GCRuleIntersection([
        cf.MaxVersionsGCRule(10),
        cf.MaxAgeGCRule(datetime.timedelta(days=90)),
    ]),
}

print('GC Policy Examples:\n')

for table_name, gc_rule in gc_examples.items():
    demo_table = instance.table(table_name)
    
    if not demo_table.exists():
        demo_table.create(column_families={BT_COLUMN_FAMILY: gc_rule})
        print(f'  Created: {table_name}')
    else:
        print(f'  Exists:  {table_name}')
    
    families = demo_table.list_column_families()
    for fam_name, fam in families.items():
        print(f'    Family "{fam_name}": {fam.gc_rule}')
    print()
````

### How GC Affects Historical Reads

GC policies directly impact what historical data is available:

- **MaxVersions(5):** Only the 5 most recent feature snapshots per entity are kept. Older snapshots are silently deleted during compaction. If you need features from 6 snapshots ago, they're gone.

- **MaxAge(90 days):** Any version older than 90 days is eligible for deletion. Training data extraction that needs features from 6 months ago will find empty cells.

- **Union (OR):** Aggressive cleanup — a version is deleted if it violates *either* rule. Keeps storage tight but limits historical depth.

- **Intersection (AND):** Conservative cleanup — a version is only deleted if it violates *both* rules. Keeps more history but uses more storage.

**Choosing a policy:**

| Use Case | Recommended Policy |
|----------|-------------------|
| Serving only (latest features) | `MaxVersionsGCRule(1)` |
| Short-term history (debugging, audit) | `MaxVersionsGCRule(10)` |
| Time-bounded retention (compliance) | `MaxAgeGCRule(days=N)` |
| Training data extraction | Intersection (keep history until both limits exceeded) |
| Cost-sensitive | Union (aggressive cleanup) |

````python
# Demonstrate how MaxVersions affects reads
# Write 8 versions to a table with MaxVersions(5)

demo_table = instance.table('gc-demo-max-versions')
demo_key = 'gc-test-entity'

# Write 8 versions with different timestamps
base_time = datetime.datetime(2024, 6, 1, 12, 0, 0, tzinfo=datetime.timezone.utc)

for i in range(8):
    ts = base_time + datetime.timedelta(hours=i)
    row = demo_table.direct_row(demo_key)
    row.set_cell(
        BT_COLUMN_FAMILY,
        'value',
        f'version_{i}'.encode('utf-8'),
        timestamp=ts,
    )
    row.commit()

print(f'Wrote 8 versions to gc-demo-max-versions (MaxVersions=5)')
print(f'  Versions written: version_0 through version_7')

# Read back — should only get 5 most recent (version_3 through version_7)
# Note: GC is asynchronous, so all 8 may still be visible immediately
row = demo_table.read_row(demo_key.encode('utf-8'))

if row:
    cells = row.cells.get(BT_COLUMN_FAMILY, {}).get(b'value', [])
    print(f'\nVersions readable: {len(cells)}')
    print(f'  (GC is asynchronous — old versions are deleted during compaction)')
    for cell in cells:
        print(f'    {cell.timestamp}  →  {cell.value.decode("utf-8")}')
else:
    print('Row not found.')
````

**Note on GC timing:** Bigtable garbage collection is asynchronous. When you write a 6th version to a MaxVersions(5) table, all 6 versions may be visible for a short time until compaction runs. Don't rely on GC for security-critical data deletion — it's a storage optimization, not a real-time constraint.

## Section 7: Trade-off Summary

| Aspect | Design A: History in Key | Design B: Cell Versioning |
|--------|------------------------|---------------------------|
| **Row key** | `entity_group#entity_id#reverse_timestamp` | `entity_group#entity_id` |
| **Latest read** | Prefix scan + limit 1 | `CellsColumnLimitFilter(1)` |
| **Historical read** | `RowRange` scan | `TimestampRangeFilter` |
| **GC** | Manual delete or MaxAge on column family | Automatic (`MaxVersionsGCRule` / `MaxAgeGCRule`) |
| **Storage** | More rows (each snapshot = separate row) | More versions per row (compact) |
| **Serving key compatibility** | Different key format than serving table | Same key — no translation needed |
| **Training data extraction** | Natural range scans across entities | Filter-based per-entity lookups |
| **Hotspotting risk** | Higher (sequential reverse timestamps) | Lower (writes spread by entity key) |
| **Row size limits** | N/A (one snapshot per row) | Risk with high-frequency entities |

### Recommendations

- **Use Design B (cell versioning)** as the default. It's simpler, uses the same key as serving, and automatic GC reduces operational burden. Most feature stores benefit from this approach.

- **Use Design A (history in key)** when you need efficient range scans across time — for example, if you frequently extract "all feature snapshots for entity X between date A and date B" for training pipelines, or if individual snapshots are large (avoiding wide rows).

- **Use BigQuery** for large-scale training data extraction. Bigtable excels at individual lookups; BigQuery excels at analytical joins across millions of rows. The right architecture uses both: BQ for batch training, Bigtable for online serving.
