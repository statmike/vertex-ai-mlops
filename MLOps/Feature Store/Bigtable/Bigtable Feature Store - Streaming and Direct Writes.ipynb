# Bigtable Feature Store — Streaming & Direct Writes

EXPORT DATA is great for batch sync, but some features change every second — click counts, session duration, real-time aggregates. For these, you [write directly to Bigtable](https://cloud.google.com/bigtable/docs/writes) using the Python client, skipping BigQuery entirely.

This notebook covers every write pattern [Bigtable](https://cloud.google.com/bigtable/docs/overview) offers: single writes, batch mutations, atomic increments, conditional mutations, and change streams for downstream sync.

## 1. Direct Single-Row Writes

The most basic write pattern: create a `DirectRow`, set one or more cells, and commit. This is the Bigtable equivalent of a SQL `INSERT` or `UPDATE` — except there's no schema. You write arbitrary bytes to any column family and qualifier combination.

The [Bigtable Python client](https://cloud.google.com/python/docs/reference/bigtable/latest) provides `table.direct_row(row_key)` to create a mutable row object. You accumulate mutations (set_cell, delete_cell, etc.) and then call `commit()` to apply them atomically.

### Write a Feature Row

Let's write a single entity's features — imagine a user whose click count, session duration, and cart value just changed.

````python
# Write a single feature row
row_key = 'user:10001'

row = table.direct_row(row_key)
row.set_cell(BT_COLUMN_FAMILY, 'click_count', b'42')
row.set_cell(BT_COLUMN_FAMILY, 'session_duration_sec', b'185')
row.set_cell(BT_COLUMN_FAMILY, 'cart_value', b'79.99')
row.set_cell(BT_COLUMN_FAMILY, 'last_category', b'electronics')
row.commit()

print(f'Wrote row: {row_key}')
print(f'  click_count = 42')
print(f'  session_duration_sec = 185')
print(f'  cart_value = 79.99')
print(f'  last_category = electronics')
````

### Read It Back

Verify the write by reading the row. The `read_row()` method returns a `PartialRowData` object whose `.cells` dict is keyed by column family and qualifier.

````python
# Read the row back
read_result = table.read_row(row_key)

if read_result:
    print(f'Row: {row_key}')
    for qualifier, cells in sorted(read_result.cells[BT_COLUMN_FAMILY].items()):
        value = cells[0].value.decode('utf-8')
        ts = cells[0].timestamp
        print(f'  {qualifier.decode("utf-8"):<25s} = {value:<15s} (ts: {ts})')
else:
    print(f'Row not found: {row_key}')
````

### Versioned Writes with Explicit Timestamps

Bigtable supports cell-level versioning. Each `set_cell()` can include a `timestamp` parameter (a `datetime.datetime` object). When you read back, you get the most recent version by default, but you can request multiple versions.

This is useful for features that change over time — you can look at the current value *and* the history.

````python
# Write three versions of a feature with explicit timestamps
row_key_versioned = 'user:10002'

# Timestamps must have microsecond granularity (truncated to milliseconds by Bigtable)
timestamps = [
    datetime.datetime(2025, 1, 15, 10, 0, 0, tzinfo=datetime.timezone.utc),
    datetime.datetime(2025, 1, 15, 11, 0, 0, tzinfo=datetime.timezone.utc),
    datetime.datetime(2025, 1, 15, 12, 0, 0, tzinfo=datetime.timezone.utc),
]

for i, ts in enumerate(timestamps):
    row = table.direct_row(row_key_versioned)
    click_val = str(10 + i * 5)  # 10, 15, 20
    row.set_cell(BT_COLUMN_FAMILY, 'click_count', click_val.encode('utf-8'), timestamp=ts)
    row.commit()
    print(f'  Version {i+1}: click_count={click_val} at {ts.isoformat()}')

# Read back all versions (our GC policy keeps 3)
read_result = table.read_row(
    row_key_versioned,
    filter_=row_filters.CellsColumnLimitFilter(3),
)

print(f'\nAll versions of click_count for {row_key_versioned}:')
for cell in read_result.cells[BT_COLUMN_FAMILY][b'click_count']:
    print(f'  value={cell.value.decode("utf-8"):<5s} timestamp={cell.timestamp}')
````

### Measure Single-Row Write Latency

How fast is a single direct write? We measure across multiple iterations to get a reliable estimate.

````python
# Benchmark single-row write latency
N_ITERATIONS = 50
latencies = []

for i in range(N_ITERATIONS):
    key = f'bench:single:{i}'
    row = table.direct_row(key)
    row.set_cell(BT_COLUMN_FAMILY, 'value', str(i).encode('utf-8'))
    
    start = time.perf_counter()
    row.commit()
    latencies.append(time.perf_counter() - start)

latencies_ms = [l * 1000 for l in latencies]
print(f'Single-row write latency ({N_ITERATIONS} iterations):')
print(f'  Min:    {min(latencies_ms):.1f} ms')
print(f'  Median: {sorted(latencies_ms)[N_ITERATIONS // 2]:.1f} ms')
print(f'  P95:    {sorted(latencies_ms)[int(N_ITERATIONS * 0.95)]:.1f} ms')
print(f'  Max:    {max(latencies_ms):.1f} ms')
print(f'  Mean:   {sum(latencies_ms) / len(latencies_ms):.1f} ms')
````

### Interpreting Single-Row Writes

Single-row writes are fast — typically under 10ms. Each `commit()` call sends one RPC to the Bigtable server, which applies the mutations atomically. Within a single row, all mutations in one `commit()` are atomic: either all cells are written or none are.

**When to use:** Single-row writes are ideal for:
- Real-time feature updates triggered by user actions (a click, a page view, a purchase)
- Low-volume updates where throughput isn't critical
- Cases where you need per-row error handling

For bulk writes, the overhead of one RPC per row adds up. That's where batch mutations come in.

---
## 2. Batch Writes (MutateRows)

When you need to write many rows at once — loading historical features, backfilling, or processing a micro-batch — the [`mutate_rows()`](https://cloud.google.com/python/docs/reference/bigtable/latest/google.cloud.bigtable.table.Table#google_cloud_bigtable_table_Table_mutate_rows) method sends all mutations in a single RPC. This amortizes network overhead across many rows.

The pattern:
1. Create `DirectRow` objects (one per row key) and set their cells
2. Collect them in a list
3. Call `table.mutate_rows(rows)` to send them all at once

### Generate and Write 1,000 Rows

````python
# Generate 1,000 rows with random features
NUM_ROWS = 1000
random.seed(42)

categories = ['electronics', 'clothing', 'home', 'sports', 'books', 'food', 'toys']

rows = []
for i in range(NUM_ROWS):
    key = f'user:{20000 + i}'
    row = table.direct_row(key)
    row.set_cell(BT_COLUMN_FAMILY, 'click_count', str(random.randint(0, 200)).encode('utf-8'))
    row.set_cell(BT_COLUMN_FAMILY, 'session_duration_sec', str(random.randint(10, 3600)).encode('utf-8'))
    row.set_cell(BT_COLUMN_FAMILY, 'cart_value', f'{random.uniform(0, 500):.2f}'.encode('utf-8'))
    row.set_cell(BT_COLUMN_FAMILY, 'last_category', random.choice(categories).encode('utf-8'))
    row.set_cell(BT_COLUMN_FAMILY, 'page_views', str(random.randint(1, 50)).encode('utf-8'))
    rows.append(row)

print(f'Prepared {len(rows)} rows for batch write')
print(f'  Row key range: user:20000 to user:{20000 + NUM_ROWS - 1}')
print(f'  Columns: click_count, session_duration_sec, cart_value, last_category, page_views')
````

````python
# Write all rows in a single batch
start = time.perf_counter()
statuses = table.mutate_rows(rows)
elapsed = time.perf_counter() - start

# Check for failures
failures = [s for s in statuses if s.code != 0]

print(f'Batch write complete:')
print(f'  Rows written:  {NUM_ROWS}')
print(f'  Time:          {elapsed:.2f}s')
print(f'  Throughput:    {NUM_ROWS / elapsed:,.0f} rows/sec')
print(f'  Failures:      {len(failures)}')

if failures:
    print(f'\nFailed entries (first 5):')
    for s in failures[:5]:
        print(f'  Code: {s.code}, Message: {s.message}')
````

### Error Handling: Partial Failures

`mutate_rows()` returns a list of status objects, one per row. Individual rows can fail (e.g., row too large, invalid column family) while others succeed. This is a **partial failure** — the batch is not all-or-nothing.

Always check the returned statuses. A code of `0` means success; any other code is a [gRPC status code](https://grpc.io/docs/guides/status-codes/).

````python
# Demonstrate error handling pattern for partial failures
def write_batch_with_retry(tbl, rows, max_retries=3):
    """
    Write a batch of rows with retry logic for partial failures.
    Returns (succeeded_count, failed_count).
    """
    pending = list(rows)
    total_succeeded = 0
    
    for attempt in range(max_retries):
        if not pending:
            break
            
        statuses = tbl.mutate_rows(pending)
        
        # Separate successes from failures
        retry_rows = []
        for row_obj, status in zip(pending, statuses):
            if status.code == 0:
                total_succeeded += 1
            else:
                retry_rows.append(row_obj)
        
        if retry_rows:
            print(f'  Attempt {attempt + 1}: {len(pending) - len(retry_rows)} succeeded, '
                  f'{len(retry_rows)} to retry')
            pending = retry_rows
        else:
            pending = []
    
    failed = len(pending)
    total_succeeded += 0  # remaining are considered failed
    
    return total_succeeded, failed

# Test with a small batch (all should succeed)
test_rows = []
for i in range(10):
    r = table.direct_row(f'retry-test:{i}')
    r.set_cell(BT_COLUMN_FAMILY, 'val', b'test')
    test_rows.append(r)

succeeded, failed = write_batch_with_retry(table, test_rows)
print(f'\nResult: {succeeded} succeeded, {failed} failed')
````

### Batch Write Throughput

Let's measure how throughput scales with batch size. Bigtable's `mutate_rows()` has a maximum of 100,000 mutations per call, but the optimal batch size depends on row size, network latency, and cluster capacity.

````python
# Measure throughput at different batch sizes
batch_sizes = [10, 50, 100, 250, 500, 1000]
throughput_results = []

for batch_size in batch_sizes:
    batch_rows = []
    for i in range(batch_size):
        r = table.direct_row(f'throughput-test:{batch_size}:{i}')
        r.set_cell(BT_COLUMN_FAMILY, 'val', f'data-{i}'.encode('utf-8'))
        batch_rows.append(r)
    
    start = time.perf_counter()
    table.mutate_rows(batch_rows)
    elapsed = time.perf_counter() - start
    
    throughput = batch_size / elapsed
    throughput_results.append({
        'batch_size': batch_size,
        'time_sec': round(elapsed, 3),
        'rows_per_sec': round(throughput),
    })
    print(f'  Batch size {batch_size:>5d}: {elapsed:.3f}s = {throughput:,.0f} rows/sec')

throughput_df = pd.DataFrame(throughput_results)
print(f'\nBatch size vs throughput:')
print(throughput_df.to_string(index=False))
````

### Interpreting Batch Write Performance

Throughput increases with batch size because each `mutate_rows()` call has a fixed overhead (network round trip, RPC setup). Larger batches amortize this overhead across more rows.

**Practical guidance:**

| Batch size | Use case |
|------------|----------|
| 1–10 | Real-time updates (low volume) |
| 50–200 | Micro-batching from a streaming pipeline |
| 500–1,000 | Bulk loading, backfills |
| >1,000 | Split into multiple calls — very large batches can hit RPC size limits |

The Bigtable client library handles chunking internally for very large batches, but keeping batch sizes under 1,000 rows gives you predictable latency and error handling.

---
## 3. Read-Modify-Write (Atomic Operations)

Some feature updates aren't "set this value" — they're "add to the current value" or "append to a log." If you implement this as read-then-write, two concurrent updates can race: both read the old value, both add 1, and you lose an increment.

Bigtable solves this with **atomic read-modify-write** operations:

| Operation | Method | Data type | Use case |
|-----------|--------|-----------|----------|
| **Increment** | `increment_cell_value()` | INT64 (8-byte big-endian) | Counters: click_count, view_count, error_count |
| **Append** | `append_cell_value()` | Bytes | Event logs, accumulating data to a cell |

These operations are performed server-side in a single RPC. No read-then-write race. No distributed locks.

### Important: INT64 Encoding

The `increment_cell_value()` method works on **8-byte big-endian signed integers**. This is different from the string encoding we used for direct writes. The value must be stored as `struct.pack('>q', value)` (big-endian signed 64-bit integer), and read back with `struct.unpack('>q', value)`.

````python
# Initialize a counter with INT64 encoding
counter_key = 'counter:clicks:page-home'

# First, write the initial value as 8-byte big-endian INT64
init_row = table.direct_row(counter_key)
init_row.set_cell(BT_COLUMN_FAMILY, 'count', struct.pack('>q', 0))
init_row.commit()

print(f'Initialized counter: {counter_key} = 0')
print(f'  Encoding: 8-byte big-endian INT64 = {struct.pack(">q", 0).hex()}')
````

````python
# Atomic increment — no race conditions
# table.row() returns an AppendRow, which supports increment_cell_value()

# Increment by 1
append_row = table.append_row(counter_key)
append_row.increment_cell_value(BT_COLUMN_FAMILY, 'count', 1)
result = append_row.commit()

print('Incremented by 1')

# Increment by 5
append_row = table.append_row(counter_key)
append_row.increment_cell_value(BT_COLUMN_FAMILY, 'count', 5)
result = append_row.commit()

print('Incremented by 5')

# Read the current value
read_result = table.read_row(counter_key)
raw = read_result.cells[BT_COLUMN_FAMILY][b'count'][0].value
current = struct.unpack('>q', raw)[0]
print(f'\nCurrent value: {current} (expected: 6)')
````

### Concurrent Increment Demo

The real power of atomic increments shows when multiple writers update the same counter simultaneously. Let's increment a counter 100 times from concurrent threads and verify the final value is exactly 100.

````python
# Concurrent increment test
concurrent_key = 'counter:concurrent-test'
N_INCREMENTS = 100
N_WORKERS = 10

# Initialize to 0
init_row = table.direct_row(concurrent_key)
init_row.set_cell(BT_COLUMN_FAMILY, 'count', struct.pack('>q', 0))
init_row.commit()

def increment_once(_):
    """Increment the counter by 1."""
    ar = table.append_row(concurrent_key)
    ar.increment_cell_value(BT_COLUMN_FAMILY, 'count', 1)
    ar.commit()

# Fire all increments concurrently
start = time.perf_counter()
with concurrent.futures.ThreadPoolExecutor(max_workers=N_WORKERS) as executor:
    list(executor.map(increment_once, range(N_INCREMENTS)))
elapsed = time.perf_counter() - start

# Read final value
read_result = table.read_row(concurrent_key)
raw = read_result.cells[BT_COLUMN_FAMILY][b'count'][0].value
final = struct.unpack('>q', raw)[0]

print(f'Concurrent increment test:')
print(f'  Workers:     {N_WORKERS}')
print(f'  Increments:  {N_INCREMENTS}')
print(f'  Final value: {final}')
print(f'  Expected:    {N_INCREMENTS}')
print(f'  Correct:     {"YES" if final == N_INCREMENTS else "NO — race condition!"}')
print(f'  Time:        {elapsed:.2f}s')
print(f'  Rate:        {N_INCREMENTS / elapsed:,.0f} increments/sec')
````

### Atomic Append

The `append_cell_value()` method atomically appends bytes to a cell's value. This is useful for building up event lists or logs within a single cell — though in practice, you'd usually prefer separate column qualifiers for individual events.

````python
# Atomic append demo
append_key = 'log:user:10001:actions'

# Write initial value
init_row = table.direct_row(append_key)
init_row.set_cell(BT_COLUMN_FAMILY, 'events', b'view_product')
init_row.commit()

# Append more events
for event in [',add_to_cart', ',begin_checkout', ',purchase']:
    ar = table.append_row(append_key)
    ar.append_cell_value(BT_COLUMN_FAMILY, 'events', event.encode('utf-8'))
    ar.commit()

# Read back the accumulated value
read_result = table.read_row(append_key)
events = read_result.cells[BT_COLUMN_FAMILY][b'events'][0].value.decode('utf-8')
print(f'Accumulated events: {events}')
print(f'Event list: {events.split(",")}')
````

### Interpreting Read-Modify-Write Operations

**Key constraints:**

- `increment_cell_value()` only works on **INT64** values (8-byte big-endian). You cannot increment a string like `"42"` — it must be `struct.pack('>q', 42)`.
- `append_cell_value()` works on raw **bytes**. There's no type awareness — it just concatenates.
- Both operations create a **new version** of the cell. With a `MaxVersionsGCRule(1)`, only the latest is kept. With higher version limits, you get a history of every mutation.

**Use case for feature stores:** Atomic increments are perfect for real-time counters — click counts, page views, error rates — that are used as ML features. A streaming pipeline can increment the counter on every event, and the feature serving layer always reads the latest value without race conditions.

## 4. Conditional Writes (CheckAndMutateRow)

Conditional writes apply mutations **only if** the current row state matches a filter. This is Bigtable's version of optimistic concurrency control — "update this row only if it currently looks like X."

The API is [`ConditionalRow`](https://cloud.google.com/python/docs/reference/bigtable/latest/google.cloud.bigtable.row.ConditionalRow):

```python
row = table.conditional_row(key, filter_=some_filter)
row.set_cell(..., state=True)    # applied if filter matches
row.set_cell(..., state=False)   # applied if filter does NOT match
result = row.commit()            # returns True if filter matched
```

### Use Case: State Transitions

Imagine a feature pipeline that processes entities through states: `pending` -> `processing` -> `complete`. Multiple workers might try to pick up the same entity. Conditional writes ensure only one worker wins.

````python
# Set up an entity in 'pending' state
state_key = 'job:feature-compute:user-batch-42'

row = table.direct_row(state_key)
row.set_cell(BT_COLUMN_FAMILY, 'status', b'pending')
row.set_cell(BT_COLUMN_FAMILY, 'created_at', b'2025-01-15T10:00:00Z')
row.commit()

print(f'Created job: {state_key}')
print(f'  status = pending')
````

````python
# Conditional write: only transition from 'pending' to 'processing'
# The filter checks if the current value of 'status' is exactly 'pending'
filter_ = row_filters.RowFilterChain(filters=[
    row_filters.FamilyNameRegexFilter(BT_COLUMN_FAMILY),
    row_filters.ColumnQualifierRegexFilter(b'status'),
    row_filters.ValueRegexFilter(b'pending'),
])

cond_row = table.conditional_row(state_key, filter_=filter_)
cond_row.set_cell(BT_COLUMN_FAMILY, 'status', b'processing', state=True)
cond_row.set_cell(BT_COLUMN_FAMILY, 'started_at', b'2025-01-15T10:00:05Z', state=True)
cond_row.set_cell(BT_COLUMN_FAMILY, 'conflict', b'another worker already claimed this job', state=False)

matched = cond_row.commit()

print(f'First claim attempt:')
print(f'  Filter matched (was pending): {matched}')

# Read back to verify
read_result = table.read_row(state_key)
status = read_result.cells[BT_COLUMN_FAMILY][b'status'][0].value.decode('utf-8')
print(f'  Current status: {status}')
````

````python
# Second attempt — should fail because status is now 'processing', not 'pending'
cond_row2 = table.conditional_row(state_key, filter_=filter_)
cond_row2.set_cell(BT_COLUMN_FAMILY, 'status', b'processing', state=True)
cond_row2.set_cell(BT_COLUMN_FAMILY, 'conflict', b'job already claimed', state=False)

matched2 = cond_row2.commit()

print(f'Second claim attempt:')
print(f'  Filter matched (was pending): {matched2}')

# Read back
read_result = table.read_row(state_key)
status = read_result.cells[BT_COLUMN_FAMILY][b'status'][0].value.decode('utf-8')
print(f'  Current status: {status}  (unchanged — second worker lost the race)')
````

### Use Case: Idempotent Feature Updates

Conditional writes are also useful for ensuring a feature update is applied exactly once. For example, only write a new feature value if the current version is older than what you're writing.

````python
# Idempotent update: only write if current version is 'v1'
feature_key = 'entity:product:SKU-1234'

# Initialize with v1 features
row = table.direct_row(feature_key)
row.set_cell(BT_COLUMN_FAMILY, 'embedding_version', b'v1')
row.set_cell(BT_COLUMN_FAMILY, 'embedding', b'[0.1, 0.2, 0.3]')
row.commit()
print('Initialized product SKU-1234 with v1 embedding')

# Upgrade to v2 only if still on v1
version_filter = row_filters.RowFilterChain(filters=[
    row_filters.FamilyNameRegexFilter(BT_COLUMN_FAMILY),
    row_filters.ColumnQualifierRegexFilter(b'embedding_version'),
    row_filters.ValueRegexFilter(b'v1'),
])

upgrade_row = table.conditional_row(feature_key, filter_=version_filter)
upgrade_row.set_cell(BT_COLUMN_FAMILY, 'embedding_version', b'v2', state=True)
upgrade_row.set_cell(BT_COLUMN_FAMILY, 'embedding', b'[0.15, 0.25, 0.35, 0.05]', state=True)

matched = upgrade_row.commit()
print(f'Upgrade v1->v2: {"applied" if matched else "skipped (not on v1)"}')

# Try upgrading again — should be skipped (already v2)
upgrade_row2 = table.conditional_row(feature_key, filter_=version_filter)
upgrade_row2.set_cell(BT_COLUMN_FAMILY, 'embedding_version', b'v2', state=True)

matched2 = upgrade_row2.commit()
print(f'Upgrade v1->v2 (retry): {"applied" if matched2 else "skipped (already v2)"}')

# Verify final state
read_result = table.read_row(feature_key)
version = read_result.cells[BT_COLUMN_FAMILY][b'embedding_version'][0].value.decode('utf-8')
emb = read_result.cells[BT_COLUMN_FAMILY][b'embedding'][0].value.decode('utf-8')
print(f'\nFinal state: version={version}, embedding={emb}')
````

### Interpreting Conditional Writes

Conditional writes are powerful but have limitations:

| Aspect | Details |
|--------|--------|
| **Scope** | Single row only — no cross-row conditions |
| **Filter** | Any Bigtable row filter (value regex, column qualifier, timestamp range, etc.) |
| **Branching** | Two branches: `on_true` (filter matches) and `on_false` (doesn't) |
| **Atomicity** | The check + mutation is atomic — no TOCTOU race |
| **Return value** | Boolean indicating which branch executed |
| **Cost** | Same as a regular write — no extra charge for the condition check |

**When to use conditional writes:**
- State machine transitions (pending -> processing -> complete)
- Version-guarded updates (only overwrite if current version matches)
- Optimistic locking without external lock services
- Idempotent retry handling in streaming pipelines

## 5. Bigtable Change Streams

[Change streams](https://cloud.google.com/bigtable/docs/change-streams-overview) emit row mutations as they happen, allowing downstream systems to react to Bigtable writes in near-real-time. This is Bigtable's answer to CDC (Change Data Capture) — you write features to Bigtable for serving, and the change stream propagates those mutations to BigQuery for analytics, Pub/Sub for event processing, or any other consumer.

### Architecture

```
App → Bigtable (source of truth for serving)
        │
        └→ Change Stream → BigQuery    (analytics)
                         → Pub/Sub     (event processing)
                         → Custom app  (any downstream system)
```

### Enable Change Streams

Change streams are enabled per-table via the admin API or gcloud. We'll show both approaches.

> **Note:** Change streams are available on Bigtable instances running production or development clusters. They add a small storage and throughput overhead for tracking mutations.

````python
# Enable change streams via gcloud
# This modifies the table to start tracking mutations
import subprocess

result = subprocess.run([
    'gcloud', 'bigtable', 'tables', 'update', BT_TABLE,
    '--instance', BT_INSTANCE,
    '--project', PROJECT_ID,
    '--change-stream-retention-period', '7d',
], capture_output=True, text=True)

if result.returncode == 0:
    print(f'Change streams enabled on table: {BT_TABLE}')
    print(f'  Retention: 7 days')
else:
    # Change streams may not be available on all instance types or API versions
    print(f'Change streams setup result:')
    print(f'  stdout: {result.stdout.strip()}')
    print(f'  stderr: {result.stderr.strip()}')
    print(f'\nNote: Change streams require Bigtable instances with the feature enabled.')
    print(f'If unavailable, the rest of this section is conceptual.')
````

### Reading Change Streams

The Python client provides `read_change_stream()` to consume mutations. Each change record contains:

| Field | Description |
|-------|-------------|
| `row_key` | The row that was mutated |
| `type_` | `SET_CELL`, `DELETE_CELLS`, `DELETE_FAMILY`, `DELETE_ROW` |
| `source_cluster_id` | Which cluster received the write |
| `commit_timestamp` | When the mutation was committed |
| `column_family` | Affected column family |
| `qualifier` | Affected column qualifier |
| `value` | The new cell value (for `SET_CELL`) |

In production, you'd run a long-lived process (Dataflow, a Kubernetes pod, a Cloud Run service) that continuously reads from the change stream and writes to downstream systems.

Here's the code pattern for reading change stream entries:

````python
# Change stream reading pattern (reference code)
#
# The ReadChangeStream API is a streaming RPC that delivers mutations
# as they occur. This code shows the usage pattern.
#
# In practice, you'd run this in a long-lived process, not a notebook.
# Google provides a Dataflow connector for production use:
#   https://cloud.google.com/bigtable/docs/change-streams-overview#dataflow

def process_change_stream_demo():
    """
    Demonstrates reading from a Bigtable change stream.
    This is a reference pattern — the actual API call requires
    the change stream to be enabled and may not work on all
    instance configurations.
    """
    from google.cloud.bigtable_v2 import BigtableClient
    from google.cloud.bigtable_v2.types import ReadChangeStreamRequest
    
    client = BigtableClient()
    table_name = f'projects/{PROJECT_ID}/instances/{BT_INSTANCE}/tables/{BT_TABLE}'
    
    request = ReadChangeStreamRequest(
        table_name=table_name,
        # Start from 5 minutes ago to catch recent writes
        start_time=datetime.datetime.now(datetime.timezone.utc) - datetime.timedelta(minutes=5),
    )
    
    try:
        stream = client.read_change_stream(request=request)
        count = 0
        for response in stream:
            if hasattr(response, 'data_change'):
                dc = response.data_change
                print(f'  Row: {dc.row_key.decode("utf-8")}')
                print(f'  Type: {dc.type_}')
                print(f'  Timestamp: {dc.commit_timestamp}')
                for chunk in dc.chunks:
                    if hasattr(chunk, 'set_cell'):
                        sc = chunk.set_cell
                        print(f'  Set: {sc.family_name}:{sc.qualifier.decode("utf-8")} '
                              f'= {sc.value.decode("utf-8", errors="replace")}')
                print()
                count += 1
                if count >= 5:  # Read only a few for demo
                    break
    except Exception as e:
        print(f'Change stream read: {e}')
        print('This is expected if change streams are not enabled or configured on this instance.')

print('Change stream reading pattern defined.')
print('Attempting to read recent changes...')
print()

try:
    process_change_stream_demo()
except Exception as e:
    print(f'Change stream not available: {e}')
    print('\nThis is expected on development instances or when the API is not enabled.')
    print('In production, use the Dataflow connector for reliable change stream processing:')
    print('  https://cloud.google.com/bigtable/docs/change-streams-overview#dataflow')
````

## 6. Dual-Write Architecture Patterns

In most ML systems, features need to live in two places:
- **Bigtable** for low-latency serving (the model reads features at inference time)
- **BigQuery** for analytics, monitoring, and retraining (data scientists query historical features)

The question is: how do you keep them in sync? There are two dominant patterns.

### Pattern 1: Fan-Out from Pub/Sub

```
Application → Pub/Sub Topic
                    │
                    ├→ Dataflow Pipeline A → Bigtable  (serving)
                    │
                    └→ Dataflow Pipeline B → BigQuery  (analytics)
                         (or BQ subscription)
```

**How it works:**
1. The application publishes feature update events to a Pub/Sub topic
2. Two independent consumers read from the topic:
   - A Dataflow pipeline that writes to Bigtable for serving
   - A Dataflow pipeline (or [BigQuery Pub/Sub subscription](https://cloud.google.com/pubsub/docs/bigquery)) that writes to BigQuery for analytics

**Advantages:**
- Pub/Sub guarantees at-least-once delivery to both consumers
- Each consumer can fail and recover independently
- Dataflow handles backpressure and scaling
- Adding a new consumer (e.g., a monitoring pipeline) is just a new subscription

**Disadvantages:**
- More infrastructure to manage (two pipelines, a Pub/Sub topic)
- Bigtable and BigQuery may be temporarily inconsistent (eventual consistency)
- Higher cost (running two streaming pipelines)

### Pattern 2: Bigtable as Source of Truth + Change Stream

```
Application → Bigtable  (source of truth)
                  │
                  └→ Change Stream → Dataflow → BigQuery  (analytics)
```

**How it works:**
1. The application writes directly to Bigtable (using the patterns from sections 1–4)
2. Bigtable's change stream captures every mutation
3. A Dataflow pipeline reads the change stream and writes to BigQuery

**Advantages:**
- Simpler — one write path, one sync pipeline
- Bigtable is the single source of truth (no split-brain risk)
- Change stream is built into Bigtable (no external message bus)

**Disadvantages:**
- Change stream adds latency to the BigQuery copy (seconds to minutes)
- Change stream records are raw mutations, not business events — you may need to reconstruct semantics
- If the Bigtable write fails, nothing flows downstream

### Comparison

| Aspect | Pattern 1: Pub/Sub Fan-Out | Pattern 2: Change Stream |
|--------|---------------------------|-------------------------|
| **Write path** | App -> Pub/Sub | App -> Bigtable |
| **Sync mechanism** | Two Pub/Sub consumers | Bigtable change stream |
| **Consistency** | Eventual (independent consumers) | Eventual (change stream lag) |
| **BQ latency** | Seconds (streaming insert) | Seconds to minutes |
| **Complexity** | Higher (Pub/Sub + 2 pipelines) | Lower (1 pipeline) |
| **Failure isolation** | BT and BQ independent | BQ depends on BT |
| **New consumers** | Add a Pub/Sub subscription | Change stream is single-consumer by default |
| **Best for** | High-volume, multi-consumer architectures | Simpler systems where BT is the primary store |

## 7. Idempotency Patterns

Network failures, retries, and duplicate messages are inevitable in distributed systems. When your streaming pipeline writes a feature update to Bigtable, that write might be retried — and you need to ensure the retry doesn't corrupt the feature value.

The problem:

```
Pipeline sends: "increment click_count by 1"
Bigtable receives and applies: click_count = 43
Network timeout — pipeline doesn't see the ACK
Pipeline retries: "increment click_count by 1" (again)
Bigtable applies again: click_count = 44  ← wrong! Should be 43.
```

Three patterns solve this problem in different ways.

### Pattern 1: Row Key as Natural Dedup

If you're writing absolute values (not increments), duplicate writes are harmless — the second write overwrites the first with the same data. Design your row key to include enough information to make duplicates idempotent.

````python
# Pattern 1: Row key as natural dedup
# Key includes entity + timestamp + event_id, so duplicates overwrite harmlessly

def write_feature_idempotent(tbl, user_id, event_id, features):
    """
    Write features using a deterministic key.
    Retries produce the same key and overwrite with the same values.
    """
    key = f'user:{user_id}:event:{event_id}'
    row = tbl.direct_row(key)
    for name, value in features.items():
        row.set_cell(BT_COLUMN_FAMILY, name, str(value).encode('utf-8'))
    row.commit()
    return key

# Write once
key = write_feature_idempotent(table, '10001', 'evt-abc-123', {
    'click_count': 42,
    'cart_value': 79.99,
})
print(f'Write 1: {key}')

# "Retry" — same key, same values — harmless overwrite
key = write_feature_idempotent(table, '10001', 'evt-abc-123', {
    'click_count': 42,
    'cart_value': 79.99,
})
print(f'Write 2 (retry): {key} — overwrites with identical data')

# Read back — correct regardless of how many times we wrote
read_result = table.read_row(key)
click = read_result.cells[BT_COLUMN_FAMILY][b'click_count'][0].value.decode('utf-8')
print(f'\nResult: click_count = {click} (correct, duplicate was harmless)')
````

### Pattern 2: Conditional Writes for State Transitions

When you need to transition state (and can't just overwrite), conditional writes prevent duplicate transitions. We covered this in Section 4, but here's the idempotency angle:

````python
# Pattern 2: Conditional writes prevent duplicate state transitions

def transition_state(tbl, key, from_state, to_state):
    """
    Atomically transition from one state to another.
    Returns True if transition happened, False if state didn't match.
    Safe to retry — once transitioned, the condition prevents re-application.
    """
    filter_ = row_filters.RowFilterChain(filters=[
        row_filters.FamilyNameRegexFilter(BT_COLUMN_FAMILY),
        row_filters.ColumnQualifierRegexFilter(b'state'),
        row_filters.ValueRegexFilter(from_state.encode('utf-8')),
    ])
    
    cond_row = tbl.conditional_row(key, filter_=filter_)
    cond_row.set_cell(BT_COLUMN_FAMILY, 'state', to_state.encode('utf-8'), state=True)
    return cond_row.commit()

# Set up initial state
state_key = 'order:ORD-9999'
row = table.direct_row(state_key)
row.set_cell(BT_COLUMN_FAMILY, 'state', b'new')
row.commit()

# Transition: new -> confirmed
result1 = transition_state(table, state_key, 'new', 'confirmed')
print(f'Transition new->confirmed: {result1}')  # True

# Retry the same transition (network hiccup)
result2 = transition_state(table, state_key, 'new', 'confirmed')
print(f'Retry new->confirmed:      {result2}')  # False — already confirmed

# Verify state
read_result = table.read_row(state_key)
state = read_result.cells[BT_COLUMN_FAMILY][b'state'][0].value.decode('utf-8')
print(f'Final state: {state}  (correct — transition applied exactly once)')
````

### Pattern 3: Event IDs as Column Qualifiers

When tracking individual events (clicks, views, actions), use the event ID as the column qualifier. Each event gets its own column, and duplicate writes overwrite the same column harmlessly.

````python
# Pattern 3: Event IDs as column qualifiers
# Each event is a separate column — duplicates overwrite the same column

events_key = 'user-events:10001'

events = [
    ('evt-001', '{"action": "view_product", "product": "SKU-100", "ts": "2025-01-15T10:00:00Z"}'),
    ('evt-002', '{"action": "add_to_cart", "product": "SKU-100", "ts": "2025-01-15T10:00:30Z"}'),
    ('evt-003', '{"action": "view_product", "product": "SKU-200", "ts": "2025-01-15T10:01:00Z"}'),
]

# Write events (each event ID is a separate column qualifier)
for event_id, payload in events:
    row = table.direct_row(events_key)
    row.set_cell(BT_COLUMN_FAMILY, event_id, payload.encode('utf-8'))
    row.commit()

# Simulate retry of evt-002
row = table.direct_row(events_key)
row.set_cell(BT_COLUMN_FAMILY, 'evt-002',
             events[1][1].encode('utf-8'))  # same payload
row.commit()
print('Retried evt-002 — overwrites the same column with identical data')

# Read all events for this user
read_result = table.read_row(events_key)
print(f'\nEvents for user 10001:')
for qualifier, cells in sorted(read_result.cells[BT_COLUMN_FAMILY].items()):
    q = qualifier.decode('utf-8')
    if q.startswith('evt-'):
        payload = json.loads(cells[0].value.decode('utf-8'))
        print(f'  {q}: {payload["action"]} on {payload["product"]}')

# Count — should be 3, not 4
event_count = sum(1 for q in read_result.cells[BT_COLUMN_FAMILY].keys()
                  if q.decode('utf-8').startswith('evt-'))
print(f'\nTotal events: {event_count} (correct — retry did not create a duplicate)')
````

### Comparing Idempotency Patterns

| Pattern | Mechanism | Works for | Limitations |
|---------|-----------|-----------|-------------|
| **Row key dedup** | Same key = overwrite | Absolute value writes | Doesn't work for increments |
| **Conditional writes** | Check-and-mutate | State transitions | Single-row only; adds latency |
| **Event ID qualifiers** | Each event = column | Event logging | Row width grows with events; may need TTL |

**Rule of thumb:** Prefer absolute value writes (Pattern 1) whenever possible. They're the simplest, fastest, and naturally idempotent. Reserve increments for true counters where you can't compute the absolute value, and use conditional writes for state machines.

## 8. Write Performance Tips

Getting the most out of Bigtable writes requires understanding how the system distributes load across its nodes. Here are the key factors that affect write throughput.

### Batch Size Optimization

We measured throughput vs batch size earlier. Let's look at the data more carefully and add per-row latency.

````python
# Extended batch size benchmark with per-row latency
batch_sizes_ext = [1, 5, 10, 25, 50, 100, 250, 500]
results_ext = []

for batch_size in batch_sizes_ext:
    # Run 3 iterations and take median
    times = []
    for trial in range(3):
        batch_rows = []
        for i in range(batch_size):
            r = table.direct_row(f'perf-test:{batch_size}:{trial}:{i}')
            r.set_cell(BT_COLUMN_FAMILY, 'f1', f'{random.random():.4f}'.encode('utf-8'))
            r.set_cell(BT_COLUMN_FAMILY, 'f2', f'{random.random():.4f}'.encode('utf-8'))
            r.set_cell(BT_COLUMN_FAMILY, 'f3', f'{random.random():.4f}'.encode('utf-8'))
            batch_rows.append(r)
        
        if batch_size == 1:
            start = time.perf_counter()
            batch_rows[0].commit()
            times.append(time.perf_counter() - start)
        else:
            start = time.perf_counter()
            table.mutate_rows(batch_rows)
            times.append(time.perf_counter() - start)
    
    median_time = sorted(times)[1]
    results_ext.append({
        'batch_size': batch_size,
        'total_ms': round(median_time * 1000, 1),
        'per_row_ms': round(median_time * 1000 / batch_size, 2),
        'rows_per_sec': round(batch_size / median_time),
    })

results_df = pd.DataFrame(results_ext)
print('Batch size optimization:')
print(results_df.to_string(index=False))
````

### Connection Pooling

The Bigtable Python client uses gRPC channels internally. By default, it creates a pool of channels to the Bigtable service. For high-throughput applications, see the [performance best practices](https://cloud.google.com/bigtable/docs/performance) documentation.

- **Reuse the client** — don't create a new `bigtable.Client()` per request. The client manages connection pooling internally.
- **One client per process** — the client is thread-safe. Share it across threads in a web server.
- **Channel count** — the default pool size is typically 4 channels. For very high throughput, you can increase it:

```python
# Increase the gRPC channel pool size for high-throughput writes
from google.cloud import bigtable
from google.api_core import client_options

options = client_options.ClientOptions(
    api_endpoint='bigtable.googleapis.com:443'
)

client = bigtable.Client(
    project=PROJECT_ID,
    admin=True,
    channel=None,  # let the client create channels
)
# The client uses grpc_gcp for channel management
```

### Row Key Distribution

Bigtable distributes data across its nodes by splitting the row key space into **tablets**. If all your writes target a narrow range of keys (a "hotspot"), they'll all hit the same node:

| Key pattern | Distribution | Problem |
|-------------|-------------|--------|
| `user:1`, `user:2`, `user:3` | Sequential | Hotspot on one node |
| `2025-01-15T10:00:00` | Timestamp prefix | All recent writes hit one node |
| `ab3f...user:1` (hash prefix) | Uniform | Well-distributed |
| `electronics#user:1` | By category | Distributes across categories |

**Rule:** Avoid monotonically increasing keys (timestamps, auto-increment IDs). Use hash prefixes, reversed timestamps, or composite keys that distribute writes evenly. This is covered in depth in Notebook 6 (Row Key Design).

### Write Throughput Summary

| Factor | Impact | Recommendation |
|--------|--------|----------------|
| **Batch size** | Amortizes RPC overhead | 50–500 rows per `mutate_rows()` call |
| **Row key distribution** | Determines load balancing across nodes | Avoid monotonic keys; use hash prefixes |
| **Cell size** | Larger cells = more network overhead | Keep cells under 10 MB; prefer many small cells |
| **Column families** | Each family is stored separately | Use few families (1–3) for write-heavy tables |
| **Connection pooling** | Reuse gRPC channels | One client per process; share across threads |
| **Cluster size** | More nodes = more write throughput | Each node handles ~10,000 writes/sec |
