# Bigtable Feature Store — Serialization

[Bigtable](https://cloud.google.com/bigtable/docs/overview) stores bytes. [BigQuery](https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types) has 20+ data types. The gap between them is the serialization problem — how you encode features determines storage cost, read latency, CPU overhead, and schema flexibility.

This notebook compares five serialization methods side-by-side, using the same data, measuring the real tradeoffs.

**Prerequisites:** Run [Notebook 0: Environment](./Bigtable%20Feature%20Store%20-%20Environment.ipynb) first.

## Section 1: The Data Type Challenge

BigQuery supports over [20 data types](https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types). Bigtable [stores every cell as raw bytes](https://cloud.google.com/bigtable/docs/overview#storage-model) — no schema, no types, no enforcement. The table below is not a theoretical exercise: the `dense_features` table in our dataset was designed to include a broad range of BigQuery types to stress-test serialization. Let's see what we're working with.

````python
# Query INFORMATION_SCHEMA to see all column types in dense_features
columns_df = bq.query(f'''
    SELECT column_name, data_type
    FROM `{DATASET_REF}.INFORMATION_SCHEMA.COLUMNS`
    WHERE table_name = 'dense_features'
    ORDER BY ordinal_position
''').to_dataframe()

print(f'dense_features has {len(columns_df)} columns:\n')
print(columns_df.to_string(index=False))
````

````python
# Classify types: BT-native (easy) vs complex (need serialization)
BT_NATIVE_TYPES = {'INT64', 'FLOAT64', 'STRING', 'BYTES', 'BOOL', 'NUMERIC', 'BIGNUMERIC'}

columns_df['category'] = columns_df['data_type'].apply(
    lambda t: 'BT-native' if t.split('(')[0].strip() in BT_NATIVE_TYPES else 'complex'
)

native_cols = columns_df[columns_df['category'] == 'BT-native']
complex_cols = columns_df[columns_df['category'] == 'complex']

print(f'BT-native types ({len(native_cols)} columns):')
print(f'  These can be stored directly as bytes with simple encoding.')
for _, row in native_cols.iterrows():
    print(f'    {row["column_name"]:<30s} {row["data_type"]}')

print(f'\nComplex types ({len(complex_cols)} columns):')
print(f'  These require a serialization strategy.')
for _, row in complex_cols.iterrows():
    print(f'    {row["column_name"]:<30s} {row["data_type"]}')
````

### BQ-to-Bigtable Type Mapping

Every BigQuery type must ultimately become bytes in Bigtable. The question is *how*:

| BigQuery Type | Bigtable Storage Strategy | Notes |
|---------------|--------------------------|-------|
| `INT64` | `struct.pack('>q', val)` or `CAST(val AS STRING)` | 8 bytes binary vs variable-length text |
| `FLOAT64` | `struct.pack('>d', val)` or `CAST(val AS STRING)` | IEEE 754 vs human-readable text |
| `STRING` | UTF-8 encode directly | Already bytes |
| `BYTES` | Store as-is | Already bytes |
| `BOOL` | `b'1'`/`b'0'` or single byte | Trivial |
| `TIMESTAMP` | Microseconds as INT64, or ISO 8601 string | Choose based on sort/range needs |
| `DATE` | Days since epoch as INT64, or `'YYYY-MM-DD'` string | Same tradeoff |
| `JSON` | Store the JSON string directly | Self-describing but verbose |
| `ARRAY<T>` | JSON array string, or repeated protobuf field | No native Bigtable array type |
| `STRUCT<...>` | JSON object, or nested protobuf message | Flatten to columns or serialize |
| `GEOGRAPHY` | GeoJSON string, or WKB bytes | WKB is compact but opaque |
| `RANGE<T>` | JSON `{"start": ..., "end": ...}` or two columns | No native range type |

The fundamental tension: **binary encodings** are compact and fast to parse but opaque to humans and debugging tools. **String encodings** are readable and debuggable but larger and slower. The right choice depends on your access patterns.

The five methods we'll compare each handle this tension differently.

### Prepare Source Data

We'll work with a 1000-entity subset of `dense_features` for consistent comparison across all methods. Let's peek at the data first.

````python
# Check how many rows are available and sample a few
row_count = bq.query(f'''
    SELECT COUNT(*) AS n FROM `{DATASET_REF}.dense_features`
''').to_dataframe().iloc[0]['n']

print(f'dense_features has {row_count:,} rows')

# Preview a few rows (scalar columns only, for display)
sample_df = bq.query(f'''
    SELECT *
    FROM `{DATASET_REF}.dense_features`
    LIMIT 3
''').to_dataframe()

print(f'\nSample (first 3 rows, all columns):')
print(sample_df.T.to_string())
````

````python
# Identify the row key column (typically the first column or an entity ID)
# We'll use the first column as the row key for all methods
ROW_KEY_COLUMN = columns_df.iloc[0]['column_name']
print(f'Row key column: {ROW_KEY_COLUMN} (type: {columns_df.iloc[0]["data_type"]})')

# Get the scalar (BT-native) column names for methods that only support scalars
SCALAR_COLUMNS = native_cols[native_cols['column_name'] != ROW_KEY_COLUMN]['column_name'].tolist()
ALL_FEATURE_COLUMNS = columns_df[columns_df['column_name'] != ROW_KEY_COLUMN]['column_name'].tolist()

print(f'Scalar feature columns: {len(SCALAR_COLUMNS)}')
print(f'All feature columns:    {len(ALL_FEATURE_COLUMNS)}')
````

## Section 2: Method 1 — Native Column Mapping

The most intuitive approach: each BigQuery column becomes a separate Bigtable **column qualifier** within the [column family](https://cloud.google.com/bigtable/docs/schema-design#column-families). The STRUCT in [`EXPORT DATA`](https://cloud.google.com/bigquery/docs/export-to-bigtable) maps directly — each STRUCT field becomes a qualifier.

**How it works:**
- Each feature is stored in its own cell under `features:<column_name>`
- EXPORT DATA with a flat STRUCT: `STRUCT(col1, col2, col3, ...) AS features`
- Reading: iterate over qualifiers, parse each by its type

**Best for:** Simple, scalar-only feature sets where you occasionally need to read individual features rather than the full row.

**Limitations:** Complex types (ARRAY, STRUCT, JSON) must be serialized to STRING before writing. Many qualifiers means more metadata overhead per row.

````python
# Build the EXPORT DATA query for native column mapping
# Each scalar column becomes a separate qualifier

BIGTABLE_URI_NATIVE = (
    f'https://bigtable.googleapis.com/projects/{PROJECT_ID}'
    f'/instances/{BT_INSTANCE}'
    f'/appProfiles/{APP_PROFILE_ID}'
    f'/tables/{BT_TABLE_NATIVE}'
)

# Build STRUCT fields for scalar columns
# EXPORT DATA to Bigtable requires all values to be STRING or BYTES
struct_fields = []
for col in SCALAR_COLUMNS:
    struct_fields.append(f'CAST({col} AS STRING) AS {col}')

struct_body = ',\n            '.join(struct_fields)

EXPORT_NATIVE_SQL = f'''
EXPORT DATA OPTIONS (
    format = 'CLOUD_BIGTABLE',
    uri = '{BIGTABLE_URI_NATIVE}',
    overwrite = true
) AS
SELECT
    CAST({ROW_KEY_COLUMN} AS STRING) AS rowkey,
    STRUCT(
            {struct_body}
    ) AS {BT_COLUMN_FAMILY}
FROM `{DATASET_REF}.dense_features`
LIMIT 1000
'''

print('Native column mapping query:')
print(EXPORT_NATIVE_SQL[:500])
if len(EXPORT_NATIVE_SQL) > 500:
    print(f'... ({len(EXPORT_NATIVE_SQL)} chars total)')
````

````python
# Execute the native column mapping export
print('Exporting with native column mapping...')
start = time.perf_counter()
job = bq.query(EXPORT_NATIVE_SQL)
job.result()
native_export_time = time.perf_counter() - start

print(f'Export complete in {native_export_time:.1f}s')
print(f'  Job ID:          {job.job_id}')
print(f'  Bytes processed: {job.total_bytes_processed:,}')
if job.slot_millis:
    print(f'  Slot time:       {job.slot_millis:,} ms')
````

````python
# Write a metadata row with column-to-type mapping
native_table = instance.table(BT_TABLE_NATIVE)

native_schema = {
    'method': 'native_column_mapping',
    'version': 'v1',
    'columns': {col: columns_df[columns_df['column_name'] == col].iloc[0]['data_type']
                for col in SCALAR_COLUMNS},
}

schema_row = native_table.direct_row('#schema')
schema_row.set_cell(
    BT_COLUMN_FAMILY, '_schema',
    json.dumps(native_schema, indent=2).encode('utf-8'),
)
schema_row.commit()
print(f'Wrote #schema row to {BT_TABLE_NATIVE}')
print(f'  Qualifiers mapped: {len(native_schema["columns"])}')
````

````python
# Read back: iterate over qualifiers, parse by type
TYPE_PARSERS = {
    'INT64': int,
    'FLOAT64': float,
    'NUMERIC': float,
    'BIGNUMERIC': float,
    'STRING': str,
    'BOOL': lambda x: x.lower() in ('true', '1'),
    'BYTES': lambda x: x.encode('utf-8'),
}

def decode_native_row(bt_row, schema):
    """Decode a native column-mapped Bigtable row."""
    features = {}
    cells = bt_row.cells.get(BT_COLUMN_FAMILY, {})
    for col_name, col_type in schema['columns'].items():
        qualifier = col_name.encode('utf-8')
        if qualifier in cells:
            raw = cells[qualifier][0].value.decode('utf-8')
            base_type = col_type.split('(')[0].strip()
            parser = TYPE_PARSERS.get(base_type, str)
            try:
                features[col_name] = parser(raw)
            except (ValueError, TypeError):
                features[col_name] = raw  # fallback to string
        else:
            features[col_name] = None
    return features

# Read a sample row
sample_rows = native_table.read_rows(
    filter_=CellsColumnLimitFilter(1),
    limit=5,
)
sample_rows.consume_all()

data_keys = [k for k in sample_rows.rows.keys() if k != b'#schema']
if data_keys:
    test_key = data_keys[0]
    test_row = native_table.read_row(test_key, filter_=CellsColumnLimitFilter(1))
    features = decode_native_row(test_row, native_schema)
    print(f'Row key: {test_key.decode("utf-8")}')
    print(f'Qualifiers: {len(features)}')
    for k, v in list(features.items())[:10]:
        print(f'  {k:<30s} = {v}')
    if len(features) > 10:
        print(f'  ... ({len(features) - 10} more)')
````

### Native Column Mapping — Analysis

**Pros:**
- Individual column reads are efficient — you can read a single qualifier without fetching the entire row
- Bigtable column filters work naturally — you can request a subset of qualifiers
- Easy to inspect in the Bigtable console — each feature is visible as a named column
- Atomic single-column updates — update one feature without touching others

**Cons:**
- Many qualifiers increase per-row metadata overhead (qualifier names stored per row)
- Complex types (ARRAY, STRUCT) must be serialized to STRING anyway
- Reading all features requires iterating over all qualifiers — more round-trips in the storage layer
- EXPORT DATA query gets unwieldy with many columns

---
## Section 3: Method 2 — JSON Casting

Serialize the entire row as a JSON string in a single Bigtable cell. BigQuery's `TO_JSON_STRING()` handles the conversion natively — including arrays, structs, timestamps, and every other type.

**How it works:**
- `TO_JSON_STRING(STRUCT(col1, col2, ...))` produces a self-describing JSON blob
- Store the JSON string in a single qualifier: `features:_json`
- Read back: `json.loads()` — trivially simple

**Best for:** Prototyping, debugging, and datasets with complex/nested types where human readability matters.

> See [`EXPORT DATA` statement reference](https://cloud.google.com/bigquery/docs/reference/standard-sql/other-statements#export_data_statement) for the full syntax and supported options.

````python
# Build the EXPORT DATA query for JSON serialization
# TO_JSON_STRING handles ALL BigQuery types including arrays and structs

BIGTABLE_URI_JSON = (
    f'https://bigtable.googleapis.com/projects/{PROJECT_ID}'
    f'/instances/{BT_INSTANCE}'
    f'/appProfiles/{APP_PROFILE_ID}'
    f'/tables/{BT_TABLE_JSON}'
)

# Build the inner STRUCT with all feature columns
all_cols_str = ', '.join(ALL_FEATURE_COLUMNS)

EXPORT_JSON_SQL = f'''
EXPORT DATA OPTIONS (
    format = 'CLOUD_BIGTABLE',
    uri = '{BIGTABLE_URI_JSON}',
    overwrite = true
) AS
SELECT
    CAST({ROW_KEY_COLUMN} AS STRING) AS rowkey,
    STRUCT(
        TO_JSON_STRING(
            STRUCT({all_cols_str})
        ) AS _json
    ) AS {BT_COLUMN_FAMILY}
FROM `{DATASET_REF}.dense_features`
LIMIT 1000
'''

print('JSON serialization query:')
print(EXPORT_JSON_SQL[:600])
if len(EXPORT_JSON_SQL) > 600:
    print(f'... ({len(EXPORT_JSON_SQL)} chars total)')
````

````python
# Execute the JSON export
print('Exporting with JSON serialization...')
start = time.perf_counter()
job = bq.query(EXPORT_JSON_SQL)
job.result()
json_export_time = time.perf_counter() - start

print(f'Export complete in {json_export_time:.1f}s')
print(f'  Job ID:          {job.job_id}')
print(f'  Bytes processed: {job.total_bytes_processed:,}')
if job.slot_millis:
    print(f'  Slot time:       {job.slot_millis:,} ms')
````

````python
# Write schema metadata (for consistency, though JSON is self-describing)
json_table = instance.table(BT_TABLE_JSON)

json_schema = {
    'method': 'json_casting',
    'version': 'v1',
    'qualifier': '_json',
    'encoding': 'TO_JSON_STRING',
    '_schema_version': 'v1',
    'columns': ALL_FEATURE_COLUMNS,
}

schema_row = json_table.direct_row('#schema')
schema_row.set_cell(
    BT_COLUMN_FAMILY, '_schema',
    json.dumps(json_schema, indent=2).encode('utf-8'),
)
schema_row.commit()
print(f'Wrote #schema row to {BT_TABLE_JSON}')
````

### JSON — Self-Description

Notice that the JSON blob embeds field names in every row. This is both a feature and a cost:

- **Feature:** No external schema needed. A developer can `json.loads()` a cell and immediately see what's inside — field names, nested structures, arrays. Debugging is trivial.
- **Cost:** Field names repeat in every row. If you have 20 columns and 1M rows, the field name overhead alone could be 100+ MB. For a 10-character field name stored 1M times, that's 10 MB per field.

You can embed a `_schema_version` key in the JSON itself for version tracking:

```sql
TO_JSON_STRING(STRUCT('v1' AS _schema_version, col1, col2, ...))
```

The reader checks `_schema_version` to know which fields to expect, enabling schema evolution without external metadata.

## Section 4: Method 3 — Pipe-Delimited Concatenation

This is the approach used in [Notebook 1](./lab_5_mlops_1_feature_store.ipynb): CONCAT all feature values with a `|` delimiter into a single cell. A separate `#schema` row stores the column order and types.

**How it works:**
- `CONCAT(CAST(col1 AS STRING), '|', CAST(col2 AS STRING), ...)` → single cell written via [`EXPORT DATA`](https://cloud.google.com/bigquery/docs/export-to-bigtable)
- Schema row stores ordered field names and types
- Read back: `split('|')` + type-specific parse

**Best for:** Simple, scalar-only feature sets where storage efficiency matters and all columns are always read together.

````python
# Build the EXPORT DATA query for pipe-delimited concatenation
# Only works reliably with scalar types

BIGTABLE_URI_CONCAT = (
    f'https://bigtable.googleapis.com/projects/{PROJECT_ID}'
    f'/instances/{BT_INSTANCE}'
    f'/appProfiles/{APP_PROFILE_ID}'
    f'/tables/{BT_TABLE_CONCAT}'
)

# Build CONCAT expression for scalar columns only
concat_parts = []
for i, col in enumerate(SCALAR_COLUMNS):
    if i > 0:
        concat_parts.append("'|'")
    concat_parts.append(f'CAST({col} AS STRING)')

concat_expr = ',\n            '.join(concat_parts)

EXPORT_CONCAT_SQL = f'''
EXPORT DATA OPTIONS (
    format = 'CLOUD_BIGTABLE',
    uri = '{BIGTABLE_URI_CONCAT}',
    overwrite = true
) AS
SELECT
    CAST({ROW_KEY_COLUMN} AS STRING) AS rowkey,
    STRUCT(
        CONCAT(
            {concat_expr}
        ) AS _values
    ) AS {BT_COLUMN_FAMILY}
FROM `{DATASET_REF}.dense_features`
LIMIT 1000
'''

print('Pipe-delimited concatenation query:')
print(EXPORT_CONCAT_SQL[:500])
if len(EXPORT_CONCAT_SQL) > 500:
    print(f'... ({len(EXPORT_CONCAT_SQL)} chars total)')
````

````python
# Execute the concat export
print('Exporting with pipe-delimited concatenation...')
start = time.perf_counter()
job = bq.query(EXPORT_CONCAT_SQL)
job.result()
concat_export_time = time.perf_counter() - start

print(f'Export complete in {concat_export_time:.1f}s')
print(f'  Job ID:          {job.job_id}')
print(f'  Bytes processed: {job.total_bytes_processed:,}')
if job.slot_millis:
    print(f'  Slot time:       {job.slot_millis:,} ms')
````

````python
# Write schema metadata
concat_table = instance.table(BT_TABLE_CONCAT)

concat_schema = {
    'method': 'pipe_delimited_concat',
    'version': 'v1',
    'delimiter': '|',
    'columns': SCALAR_COLUMNS,
    'types': {col: columns_df[columns_df['column_name'] == col].iloc[0]['data_type']
              for col in SCALAR_COLUMNS},
}

schema_row = concat_table.direct_row('#schema')
schema_row.set_cell(
    BT_COLUMN_FAMILY, '_schema',
    json.dumps(concat_schema, indent=2).encode('utf-8'),
)
schema_row.commit()
print(f'Wrote #schema row to {BT_TABLE_CONCAT}')
print(f'  Columns: {len(SCALAR_COLUMNS)}')
print(f'  Delimiter: |')
````

````python
# Read back: split + parse by type
def decode_concat_row(raw_value, schema):
    """Decode a pipe-delimited feature string using the schema."""
    values = raw_value.split(schema['delimiter'])
    features = {}
    for col, val in zip(schema['columns'], values):
        base_type = schema['types'][col].split('(')[0].strip()
        parser = TYPE_PARSERS.get(base_type, str)
        try:
            features[col] = parser(val)
        except (ValueError, TypeError):
            features[col] = val
    return features

# Read a sample
sample_rows = concat_table.read_rows(
    filter_=CellsColumnLimitFilter(1), limit=5,
)
sample_rows.consume_all()
data_keys_concat = [k for k in sample_rows.rows.keys() if k != b'#schema']

if data_keys_concat:
    test_key = data_keys_concat[0]
    test_row = concat_table.read_row(test_key, filter_=CellsColumnLimitFilter(1))
    raw_val = test_row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
    features = decode_concat_row(raw_val, concat_schema)

    print(f'Row key: {test_key.decode("utf-8")}')
    print(f'Raw value size: {len(raw_val)} bytes')
    print(f'Raw value (first 200 chars): {raw_val[:200]}')
    print(f'\nParsed features (first 10):')
    for k, v in list(features.items())[:10]:
        print(f'  {k:<30s} = {v}')
````

### The Delimiter Trap

Pipe-delimited concatenation works perfectly for numeric features. But what happens when a STRING feature contains the delimiter character?

```
Stored:   42|hello|world|3.14
Parsed:   ['42', 'hello', 'world', '3.14'] → 4 fields ✔

Stored:   42|hello|world|it's a pipe: | right?|3.14
Parsed:   ['42', 'hello', 'world', "it's a pipe: ", ' right?', '3.14'] → 6 fields ✘
```

The parser sees 6 fields instead of 4 — **silent data corruption**. The feature values are wrong, and there's no error to catch it.

**Mitigation strategies:**

| Strategy | How | Trade-off |
|----------|-----|----------|
| **Rare delimiter** | Use `\x01` (SOH) or `\x1f` (unit separator) | Still vulnerable if data contains the byte |
| **Escaping** | Replace `\|` with `\\\|` in values, unescape on read | Adds complexity to both writer and reader |
| **Length-prefix** | Store `len:value` for each field | More complex but unambiguous |
| **Don't use delimiters** | Use JSON or Protobuf instead | Different method entirely |

For pure numeric feature sets (like our behavioral features), the delimiter trap is not a concern. For mixed-type feature sets with free-text fields, avoid this method.

## Section 5: Method 4 — Protocol Buffers

Protocol Buffers (protobuf) are Google's language-neutral, platform-neutral binary serialization format. They produce the most compact encoding and the fastest parsing, making them the standard for production feature stores.

**How it works:**
- Define a `.proto` schema describing the feature fields
- On write: serialize features to binary protobuf bytes, store in a single cell
- On read: deserialize with the protobuf library
- Store the serialized `FileDescriptorSet` in a `#schema` row for self-description

**Best for:** Production systems where storage cost and parsing speed are critical.

**Note:** EXPORT DATA doesn't support protobuf encoding natively, so we [write via the Bigtable Python client](https://cloud.google.com/python/docs/reference/bigtable/latest) after querying the data from BigQuery.

````python
# Build a protobuf schema dynamically from the BigQuery column types
# We use google.protobuf.descriptor_pool to create descriptors at runtime
# rather than compiling a .proto file

from google.protobuf import descriptor_pb2
from google.protobuf import descriptor_pool
from google.protobuf import descriptor
from google.protobuf import message_factory
from google.protobuf import symbol_database

# Map BigQuery types to protobuf field types
BQ_TO_PROTO_TYPE = {
    'INT64': descriptor_pb2.FieldDescriptorProto.TYPE_INT64,
    'FLOAT64': descriptor_pb2.FieldDescriptorProto.TYPE_DOUBLE,
    'STRING': descriptor_pb2.FieldDescriptorProto.TYPE_STRING,
    'BOOL': descriptor_pb2.FieldDescriptorProto.TYPE_BOOL,
    'BYTES': descriptor_pb2.FieldDescriptorProto.TYPE_BYTES,
    'NUMERIC': descriptor_pb2.FieldDescriptorProto.TYPE_STRING,  # store as string
    'BIGNUMERIC': descriptor_pb2.FieldDescriptorProto.TYPE_STRING,
    'TIMESTAMP': descriptor_pb2.FieldDescriptorProto.TYPE_STRING,
    'DATE': descriptor_pb2.FieldDescriptorProto.TYPE_STRING,
    'JSON': descriptor_pb2.FieldDescriptorProto.TYPE_STRING,
}

# Build a FileDescriptorProto for our features
file_descriptor_proto = descriptor_pb2.FileDescriptorProto(
    name='features.proto',
    package='featurestore',
    syntax='proto3',
    message_type=[
        descriptor_pb2.DescriptorProto(
            name='FeatureRow',
            field=[
                descriptor_pb2.FieldDescriptorProto(
                    name=col,
                    number=i + 1,
                    type=BQ_TO_PROTO_TYPE.get(
                        columns_df[columns_df['column_name'] == col].iloc[0]['data_type'].split('(')[0].strip(),
                        descriptor_pb2.FieldDescriptorProto.TYPE_STRING  # fallback
                    ),
                    label=descriptor_pb2.FieldDescriptorProto.LABEL_OPTIONAL,
                )
                for i, col in enumerate(SCALAR_COLUMNS)
            ],
        ),
    ],
)

# Serialize the FileDescriptorSet for storage
file_descriptor_set = descriptor_pb2.FileDescriptorSet(
    file=[file_descriptor_proto]
)
proto_schema_bytes = file_descriptor_set.SerializeToString()

print(f'Proto schema:')
print(f'  Message: featurestore.FeatureRow')
print(f'  Fields:  {len(SCALAR_COLUMNS)}')
print(f'  Schema size: {len(proto_schema_bytes)} bytes (serialized FileDescriptorSet)')

# Show the proto definition
print(f'\nEquivalent .proto definition:')
print(f'syntax = "proto3";')
print(f'package featurestore;')
print(f'message FeatureRow {{')
for i, col in enumerate(SCALAR_COLUMNS[:10]):
    base_type = columns_df[columns_df['column_name'] == col].iloc[0]['data_type'].split('(')[0].strip()
    proto_type_map = {'INT64': 'int64', 'FLOAT64': 'double', 'STRING': 'string', 'BOOL': 'bool'}
    proto_t = proto_type_map.get(base_type, 'string')
    print(f'  {proto_t} {col} = {i + 1};')
if len(SCALAR_COLUMNS) > 10:
    print(f'  // ... {len(SCALAR_COLUMNS) - 10} more fields')
print(f'}}')
````

````python
# Create the protobuf message class from the descriptor
pool = descriptor_pool.DescriptorPool()
pool.Add(file_descriptor_proto)
msg_descriptor = pool.FindMessageTypeByName('featurestore.FeatureRow')
factory = message_factory.MessageFactory(pool=pool)
FeatureRow = factory.GetPrototype(msg_descriptor)

print(f'Created FeatureRow message class: {FeatureRow.DESCRIPTOR.name}')
print(f'  Fields: {len(FeatureRow.DESCRIPTOR.fields)}')
````

````python
# Query data from BigQuery and write as protobuf to Bigtable
print('Querying data from BigQuery...')
scalar_cols_str = ', '.join(SCALAR_COLUMNS)
proto_df = bq.query(f'''
    SELECT CAST({ROW_KEY_COLUMN} AS STRING) AS entity_id, {scalar_cols_str}
    FROM `{DATASET_REF}.dense_features`
    LIMIT 1000
''').to_dataframe()

print(f'  Rows: {len(proto_df)}')

# Serialize each row as protobuf and write to Bigtable
proto_table = instance.table(BT_TABLE_PROTO)

# Write schema row with the FileDescriptorSet
schema_row = proto_table.direct_row('#schema')
schema_row.set_cell(
    BT_COLUMN_FAMILY, '_proto_schema',
    proto_schema_bytes,  # binary protobuf schema
)
schema_row.set_cell(
    BT_COLUMN_FAMILY, '_schema',
    json.dumps({
        'method': 'protocol_buffers',
        'version': 'v1',
        'message_name': 'featurestore.FeatureRow',
        'qualifier': '_proto',
        'columns': SCALAR_COLUMNS,
    }, indent=2).encode('utf-8'),
)
schema_row.commit()
print(f'Wrote #schema row with FileDescriptorSet to {BT_TABLE_PROTO}')

# Write data rows in batches
proto_sizes = []
BATCH_SIZE = 100
start = time.perf_counter()

for batch_start in range(0, len(proto_df), BATCH_SIZE):
    batch = proto_df.iloc[batch_start:batch_start + BATCH_SIZE]
    rows_to_write = []

    for _, pdf_row in batch.iterrows():
        msg = FeatureRow()
        for col in SCALAR_COLUMNS:
            val = pdf_row[col]
            if pd.notna(val):
                field = msg.DESCRIPTOR.fields_by_name[col]
                if field.type == descriptor_pb2.FieldDescriptorProto.TYPE_INT64:
                    setattr(msg, col, int(val))
                elif field.type == descriptor_pb2.FieldDescriptorProto.TYPE_DOUBLE:
                    setattr(msg, col, float(val))
                elif field.type == descriptor_pb2.FieldDescriptorProto.TYPE_BOOL:
                    setattr(msg, col, bool(val))
                else:
                    setattr(msg, col, str(val))

        proto_bytes = msg.SerializeToString()
        proto_sizes.append(len(proto_bytes))

        bt_row = proto_table.direct_row(str(pdf_row['entity_id']))
        bt_row.set_cell(BT_COLUMN_FAMILY, '_proto', proto_bytes)
        rows_to_write.append(bt_row)

    # Write batch
    statuses = proto_table.mutate_rows(rows_to_write)
    failed = sum(1 for s in statuses if s.code != 0)
    if failed:
        print(f'  Batch {batch_start}: {failed} failures')

proto_write_time = time.perf_counter() - start

print(f'\nWrote {len(proto_df)} rows as protobuf in {proto_write_time:.1f}s')
print(f'  Avg proto size: {np.mean(proto_sizes):.0f} bytes')
print(f'  Min proto size: {np.min(proto_sizes)} bytes')
print(f'  Max proto size: {np.max(proto_sizes)} bytes')
````

````python
# Read back: deserialize with protobuf library
sample_rows = proto_table.read_rows(
    filter_=CellsColumnLimitFilter(1), limit=5,
)
sample_rows.consume_all()
data_keys_proto = [k for k in sample_rows.rows.keys() if k != b'#schema']

if data_keys_proto:
    test_key = data_keys_proto[0]
    test_row = proto_table.read_row(test_key, filter_=CellsColumnLimitFilter(1))
    proto_bytes = test_row.cells[BT_COLUMN_FAMILY][b'_proto'][0].value

    # Deserialize
    msg = FeatureRow()
    msg.ParseFromString(proto_bytes)

    # Convert to dict
    features = {}
    for field in msg.DESCRIPTOR.fields:
        features[field.name] = getattr(msg, field.name)

    print(f'Row key: {test_key.decode("utf-8")}')
    print(f'Proto size: {len(proto_bytes)} bytes')
    print(f'\nDeserialized features (first 10):')
    for k, v in list(features.items())[:10]:
        print(f'  {k:<30s} = {v}')
    if len(features) > 10:
        print(f'  ... ({len(features) - 10} more)')
````

### Protocol Buffers — Self-Describing via FileDescriptorSet

We stored the serialized `FileDescriptorSet` in the `#schema` row alongside the human-readable JSON schema. This allows any client to reconstruct the message class at runtime:

```python
# Reader side (no .proto file needed):
from google.protobuf import descriptor_pb2, descriptor_pool, message_factory

# Read the schema row
schema_bytes = table.read_row('#schema').cells['features'][b'_proto_schema'][0].value
fds = descriptor_pb2.FileDescriptorSet()
fds.ParseFromString(schema_bytes)

# Rebuild the message class
pool = descriptor_pool.DescriptorPool()
for f in fds.file:
    pool.Add(f)
desc = pool.FindMessageTypeByName('featurestore.FeatureRow')
FeatureRow = message_factory.MessageFactory(pool=pool).GetPrototype(desc)

# Deserialize any data row
msg = FeatureRow()
msg.ParseFromString(proto_bytes)
```

This eliminates the need to distribute compiled `.proto` files or keep reader code in sync with the writer. The schema travels with the data.

## Section 6: Method 5 — Hybrid

The hybrid approach combines the best of native columns and protobuf:

- **Top-N most-accessed features** get their own column qualifiers (fast [individual reads](https://cloud.google.com/bigtable/docs/reads))
- **Remaining features** are packed into a protobuf blob (compact bulk storage)
- A **manifest column** describes which features are where

This is the approach used by production feature stores that serve both real-time lookups ("give me just the spending_ratio for this user") and batch scoring ("give me all features").

**Best for:** Mixed access patterns where some features are read individually and others only during batch scoring.

````python
# Define the split: top-10 most-accessed features as native columns,
# the rest as a protobuf blob
# In practice, you'd determine this from access logs.
# Here we pick the first 10 scalar columns as the "hot" features.

HOT_FEATURES = SCALAR_COLUMNS[:10] if len(SCALAR_COLUMNS) >= 10 else SCALAR_COLUMNS
COLD_FEATURES = SCALAR_COLUMNS[10:] if len(SCALAR_COLUMNS) > 10 else []

print(f'Hot features ({len(HOT_FEATURES)}) — individual column qualifiers:')
for col in HOT_FEATURES:
    print(f'  features:{col}')

print(f'\nCold features ({len(COLD_FEATURES)}) — packed into protobuf blob:')
for col in COLD_FEATURES:
    print(f'  {col}')
if not COLD_FEATURES:
    print(f'  (none — all features are hot)')

print(f'\nManifest qualifier: features:_manifest')
````

````python
# Build protobuf schema for cold features only
if COLD_FEATURES:
    cold_file_descriptor_proto = descriptor_pb2.FileDescriptorProto(
        name='cold_features.proto',
        package='featurestore',
        syntax='proto3',
        message_type=[
            descriptor_pb2.DescriptorProto(
                name='ColdFeatureRow',
                field=[
                    descriptor_pb2.FieldDescriptorProto(
                        name=col,
                        number=i + 1,
                        type=BQ_TO_PROTO_TYPE.get(
                            columns_df[columns_df['column_name'] == col].iloc[0]['data_type'].split('(')[0].strip(),
                            descriptor_pb2.FieldDescriptorProto.TYPE_STRING
                        ),
                        label=descriptor_pb2.FieldDescriptorProto.LABEL_OPTIONAL,
                    )
                    for i, col in enumerate(COLD_FEATURES)
                ],
            ),
        ],
    )

    cold_pool = descriptor_pool.DescriptorPool()
    cold_pool.Add(cold_file_descriptor_proto)
    cold_msg_descriptor = cold_pool.FindMessageTypeByName('featurestore.ColdFeatureRow')
    cold_factory = message_factory.MessageFactory(pool=cold_pool)
    ColdFeatureRow = cold_factory.GetPrototype(cold_msg_descriptor)
    print(f'Created ColdFeatureRow message class ({len(COLD_FEATURES)} fields)')
else:
    ColdFeatureRow = None
    print('No cold features — all features stored as native columns')
````

````python
# Query data and write in hybrid format
print('Querying data from BigQuery for hybrid write...')
scalar_cols_str = ', '.join(SCALAR_COLUMNS)
hybrid_df = bq.query(f'''
    SELECT CAST({ROW_KEY_COLUMN} AS STRING) AS entity_id, {scalar_cols_str}
    FROM `{DATASET_REF}.dense_features`
    LIMIT 1000
''').to_dataframe()

print(f'  Rows: {len(hybrid_df)}')

hybrid_table = instance.table(BT_TABLE_HYBRID)

# Write schema/manifest row
manifest = {
    'method': 'hybrid',
    'version': 'v1',
    'hot_features': HOT_FEATURES,
    'cold_features': COLD_FEATURES,
    'hot_encoding': 'string_per_qualifier',
    'cold_encoding': 'protobuf',
    'cold_qualifier': '_cold_proto',
    'cold_message': 'featurestore.ColdFeatureRow',
}

schema_row = hybrid_table.direct_row('#schema')
schema_row.set_cell(
    BT_COLUMN_FAMILY, '_manifest',
    json.dumps(manifest, indent=2).encode('utf-8'),
)
if COLD_FEATURES:
    cold_fds = descriptor_pb2.FileDescriptorSet(file=[cold_file_descriptor_proto])
    schema_row.set_cell(
        BT_COLUMN_FAMILY, '_cold_proto_schema',
        cold_fds.SerializeToString(),
    )
schema_row.commit()
print(f'Wrote #schema/manifest to {BT_TABLE_HYBRID}')

# Write data rows
hybrid_sizes = []
start = time.perf_counter()

for batch_start in range(0, len(hybrid_df), BATCH_SIZE):
    batch = hybrid_df.iloc[batch_start:batch_start + BATCH_SIZE]
    rows_to_write = []

    for _, pdf_row in batch.iterrows():
        bt_row = hybrid_table.direct_row(str(pdf_row['entity_id']))
        row_size = 0

        # Hot features: individual qualifiers
        for col in HOT_FEATURES:
            val = pdf_row[col]
            val_str = '' if pd.isna(val) else str(val)
            val_bytes = val_str.encode('utf-8')
            bt_row.set_cell(BT_COLUMN_FAMILY, col, val_bytes)
            row_size += len(val_bytes) + len(col)

        # Cold features: protobuf blob
        if COLD_FEATURES and ColdFeatureRow:
            msg = ColdFeatureRow()
            for col in COLD_FEATURES:
                val = pdf_row[col]
                if pd.notna(val):
                    field = msg.DESCRIPTOR.fields_by_name[col]
                    if field.type == descriptor_pb2.FieldDescriptorProto.TYPE_INT64:
                        setattr(msg, col, int(val))
                    elif field.type == descriptor_pb2.FieldDescriptorProto.TYPE_DOUBLE:
                        setattr(msg, col, float(val))
                    elif field.type == descriptor_pb2.FieldDescriptorProto.TYPE_BOOL:
                        setattr(msg, col, bool(val))
                    else:
                        setattr(msg, col, str(val))
            cold_bytes = msg.SerializeToString()
            bt_row.set_cell(BT_COLUMN_FAMILY, '_cold_proto', cold_bytes)
            row_size += len(cold_bytes)

        hybrid_sizes.append(row_size)
        rows_to_write.append(bt_row)

    statuses = hybrid_table.mutate_rows(rows_to_write)

hybrid_write_time = time.perf_counter() - start

print(f'\nWrote {len(hybrid_df)} rows in hybrid format in {hybrid_write_time:.1f}s')
print(f'  Avg row size: {np.mean(hybrid_sizes):.0f} bytes')
print(f'  Hot qualifiers per row: {len(HOT_FEATURES)}')
if COLD_FEATURES:
    print(f'  Cold proto qualifier: _cold_proto')
````

````python
# Read back: combine hot qualifiers + cold protobuf
def decode_hybrid_row(bt_row, manifest):
    """Decode a hybrid-encoded Bigtable row."""
    features = {}
    cells = bt_row.cells.get(BT_COLUMN_FAMILY, {})

    # Hot features: read individual qualifiers
    for col in manifest['hot_features']:
        qualifier = col.encode('utf-8')
        if qualifier in cells:
            raw = cells[qualifier][0].value.decode('utf-8')
            # Simple type inference: try int, then float, then string
            try:
                features[col] = int(raw)
            except ValueError:
                try:
                    features[col] = float(raw)
                except ValueError:
                    features[col] = raw
        else:
            features[col] = None

    # Cold features: deserialize protobuf
    cold_qualifier = manifest.get('cold_qualifier', '_cold_proto').encode('utf-8')
    if cold_qualifier in cells and ColdFeatureRow:
        msg = ColdFeatureRow()
        msg.ParseFromString(cells[cold_qualifier][0].value)
        for field in msg.DESCRIPTOR.fields:
            features[field.name] = getattr(msg, field.name)

    return features

# Read a sample
sample_rows = hybrid_table.read_rows(
    filter_=CellsColumnLimitFilter(1), limit=5,
)
sample_rows.consume_all()
data_keys_hybrid = [k for k in sample_rows.rows.keys() if k != b'#schema']

if data_keys_hybrid:
    test_key = data_keys_hybrid[0]
    test_row = hybrid_table.read_row(test_key, filter_=CellsColumnLimitFilter(1))
    features = decode_hybrid_row(test_row, manifest)

    print(f'Row key: {test_key.decode("utf-8")}')
    print(f'Total features: {len(features)}')
    print(f'\nHot features (individual qualifiers):')
    for col in HOT_FEATURES[:5]:
        print(f'  {col:<30s} = {features.get(col)}')
    if len(HOT_FEATURES) > 5:
        print(f'  ... ({len(HOT_FEATURES) - 5} more hot features)')
    if COLD_FEATURES:
        print(f'\nCold features (from protobuf blob):')
        for col in COLD_FEATURES[:5]:
            print(f'  {col:<30s} = {features.get(col)}')
        if len(COLD_FEATURES) > 5:
            print(f'  ... ({len(COLD_FEATURES) - 5} more cold features)')
````

### Hybrid — Access Pattern Optimization

The hybrid approach shines when features have different access patterns:

| Feature Group | Access Pattern | Storage | Why |
|--------------|----------------|---------|-----|
| Hot (top-10) | Individual reads, streaming updates | Native qualifiers | Fast per-feature reads, atomic updates |
| Cold (rest) | Batch reads during scoring | Protobuf blob | Compact, fast bulk decode |

**Real-world example:** In a fraud detection system:
- **Hot features** (`txn_count_5min`, `spending_ratio`) change with every transaction and are read individually by the streaming pipeline to make fast decisions
- **Cold features** (`account_age_days`, `lifetime_orders`) change infrequently and are only read during full model scoring

The manifest column (`features:_manifest`) tells the reader which features live where, enabling the same client to handle both access patterns without hardcoding.

## Section 7: The Tax Analysis

Now we compare all five methods on the same 1000-entity subset across four dimensions:

1. **Storage size** — bytes per entity
2. **Write time** — how long the export/write took
3. **Read latency** — time to read + decode 100 entities
4. **CPU time** — client-side parsing overhead (decode only, no network)

### Measure Storage Size

````python
# Measure storage size per entity for each method
def measure_row_sizes(table_obj, n_rows=100):
    """Read n_rows and measure total bytes per row."""
    rows = table_obj.read_rows(
        filter_=CellsColumnLimitFilter(1), limit=n_rows + 1,
    )
    rows.consume_all()

    sizes = []
    for key, row_data in rows.rows.items():
        if key == b'#schema':
            continue
        row_bytes = 0
        for family, qualifiers in row_data.cells.items():
            for qualifier, cells in qualifiers.items():
                for cell in cells:
                    row_bytes += len(cell.value)
                    row_bytes += len(qualifier)  # qualifier name overhead
        sizes.append(row_bytes)
    return sizes

print('Measuring storage sizes...')
native_sizes = measure_row_sizes(native_table)
json_sizes = measure_row_sizes(json_table)
concat_sizes = measure_row_sizes(concat_table)
proto_read_sizes = measure_row_sizes(proto_table)
hybrid_read_sizes = measure_row_sizes(hybrid_table)

size_results = {
    'Native Columns': np.mean(native_sizes),
    'JSON': np.mean(json_sizes),
    'Pipe-Delimited': np.mean(concat_sizes),
    'Protobuf': np.mean(proto_read_sizes),
    'Hybrid': np.mean(hybrid_read_sizes),
}

print(f'\nAverage bytes per entity:')
for method, avg_size in sorted(size_results.items(), key=lambda x: x[1]):
    print(f'  {method:<20s} {avg_size:>8.0f} bytes')
````

### Measure Read Latency

````python
# Measure read + decode latency for 100 entities per method
N_READ = 100

def benchmark_reads(table_obj, decode_fn, qualifier_or_fn, n=N_READ):
    """Read n rows and measure total time (network + decode)."""
    # Get row keys first
    rows = table_obj.read_rows(filter_=CellsColumnLimitFilter(1), limit=n + 1)
    rows.consume_all()
    keys = [k for k in rows.rows.keys() if k != b'#schema'][:n]

    # Measure read + decode
    start = time.perf_counter()
    for key in keys:
        row = table_obj.read_row(key, filter_=CellsColumnLimitFilter(1))
        decode_fn(row)
    total = time.perf_counter() - start

    # Measure decode-only (CPU time) using cached data
    # Re-read all rows first, then time just the decode
    cached_rows = []
    for key in keys[:min(20, len(keys))]:
        row = table_obj.read_row(key, filter_=CellsColumnLimitFilter(1))
        cached_rows.append(row)

    start_cpu = time.perf_counter()
    for _ in range(5):  # repeat for more stable measurement
        for row in cached_rows:
            decode_fn(row)
    cpu_time = (time.perf_counter() - start_cpu) / 5

    return total, cpu_time, len(keys)

# Decode functions for each method
def decode_native(row):
    return decode_native_row(row, native_schema)

def decode_json_method(row):
    raw = row.cells[BT_COLUMN_FAMILY][b'_json'][0].value.decode('utf-8')
    return json.loads(raw)

def decode_concat_method(row):
    raw = row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
    return decode_concat_row(raw, concat_schema)

def decode_proto_method(row):
    proto_bytes = row.cells[BT_COLUMN_FAMILY][b'_proto'][0].value
    msg = FeatureRow()
    msg.ParseFromString(proto_bytes)
    return {f.name: getattr(msg, f.name) for f in msg.DESCRIPTOR.fields}

def decode_hybrid_method(row):
    return decode_hybrid_row(row, manifest)

print(f'Benchmarking read + decode for {N_READ} entities per method...\n')

results = {}
for method_name, tbl, decode_fn in [
    ('Native Columns', native_table, decode_native),
    ('JSON', json_table, decode_json_method),
    ('Pipe-Delimited', concat_table, decode_concat_method),
    ('Protobuf', proto_table, decode_proto_method),
    ('Hybrid', hybrid_table, decode_hybrid_method),
]:
    total_time, cpu_time, n_rows = benchmark_reads(tbl, decode_fn, None)
    per_row_ms = (total_time / n_rows) * 1000
    per_row_cpu_us = (cpu_time / min(20, n_rows)) * 1_000_000
    results[method_name] = {
        'total_time': total_time,
        'per_row_ms': per_row_ms,
        'cpu_per_row_us': per_row_cpu_us,
        'n_rows': n_rows,
    }
    print(f'  {method_name:<20s}  read: {per_row_ms:.1f} ms/row  decode: {per_row_cpu_us:.0f} us/row  ({n_rows} rows)')
````

````python
# Build comparison table
comparison_data = []
for method in ['Native Columns', 'JSON', 'Pipe-Delimited', 'Protobuf', 'Hybrid']:
    comparison_data.append({
        'Method': method,
        'Avg Bytes/Entity': f'{size_results[method]:,.0f}',
        'Read Latency (ms/row)': f'{results[method]["per_row_ms"]:.1f}',
        'CPU Decode (us/row)': f'{results[method]["cpu_per_row_us"]:.0f}',
        'Handles Complex Types': 'Partial' if method in ('Native Columns', 'Pipe-Delimited') else 'Yes' if method in ('JSON', 'Protobuf') else 'Partial',
        'Individual Field Read': 'Yes' if method in ('Native Columns', 'Hybrid') else 'No',
    })

comparison_df = pd.DataFrame(comparison_data)
print('Serialization Method Comparison:')
print('=' * 100)
print(comparison_df.to_string(index=False))
````

### Recommendation Matrix

| Scenario | Recommended Method | Why |
|----------|-------------------|-----|
| **Prototype / debugging** | JSON | Human-readable, self-describing, handles all types |
| **Simple numeric features, batch ML** | Pipe-Delimited | Compact, simple, fast — the NB1 approach |
| **Production feature store, high QPS** | Protobuf | Smallest storage, fastest decode, strongly typed |
| **Mixed access patterns** | Hybrid | Best of both worlds — fast individual reads + compact bulk |
| **Many columns, occasional field reads** | Native Columns | Filter individual qualifiers without reading the full row |
| **Complex/nested types** | JSON or Protobuf | Pipe-delimited and native columns cannot handle arrays/structs natively |

### The Hidden Cost: Schema Evolution

Storage and latency are easy to measure. The harder cost is **what happens when you add a feature or change a type**:

| Method | Schema Change Impact |
|--------|---------------------|
| **Native Columns** | Add a qualifier. Old rows don't have it (NULL). No backfill needed unless the client requires it. |
| **JSON** | Add a key to the JSON. Old rows lack the key — `dict.get(key, default)` handles it. |
| **Pipe-Delimited** | Append to the end of the CONCAT and update the schema row. Old rows have fewer fields — the decoder must handle variable-length splits. |
| **Protobuf** | Add a field with a new field number. Backward compatible by design — old messages simply don't have the field. |
| **Hybrid** | Move features between hot and cold. Requires manifest update and potentially backfill. |

Protobuf wins on schema evolution because backward/forward compatibility is a core design goal of the format. JSON is a close second thanks to its self-describing nature.

> For more on designing effective Bigtable schemas, see the [schema design best practices](https://cloud.google.com/bigtable/docs/schema-design) documentation.

## Summary

We compared five serialization strategies for storing BigQuery features in Bigtable:

| Method | Storage | Parse Speed | Schema Evolution | Best For |
|--------|---------|-------------|-----------------|----------|
| **Native Columns** | Medium | Fast (per-field) | Easy (add qualifier) | Individual field access |
| **JSON** | Largest | Slowest | Easy (add key) | Prototyping, complex types |
| **Pipe-Delimited** | Small | Fast | Fragile (order-dependent) | Simple numeric features |
| **Protobuf** | Smallest | Fastest | Best (field numbers) | Production at scale |
| **Hybrid** | Medium | Mixed | Moderate (manifest update) | Mixed access patterns |

### Key Takeaways

1. **There is no universally best method.** The right choice depends on your access patterns, data types, and operational complexity budget.

2. **Start with JSON or pipe-delimited** for prototyping. They're simple, debuggable, and get you to a working feature store quickly.

3. **Graduate to protobuf** when storage costs or parse latency become bottlenecks. The FileDescriptorSet-in-schema pattern keeps it self-describing.

4. **Use hybrid** when you have distinct hot/cold access patterns and the operational complexity is justified by the performance gains.

5. **The delimiter trap is real.** Never use pipe-delimited concatenation with free-text features.
