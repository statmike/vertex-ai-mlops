## Section 1: Row Key Anatomy

Bigtable stores rows in **lexicographic order** of their row key. The row key is an arbitrary byte string, but in practice it is almost always a human-readable string with a separator character.

Key properties:

| Property | Description |
|----------|-------------|
| **Ordering** | Rows are sorted by raw bytes, not numeric value. `"9"` sorts after `"10"` because `'9' > '1'` in ASCII. |
| **Uniqueness** | Each row key must be unique within a table. Duplicate writes overwrite cells. |
| **Length** | Shorter keys = less storage overhead. But keys must be unique and descriptive. |
| **Encoding** | UTF-8 strings are common. Binary keys are possible but harder to debug. |

### Our Compound Key: `{entity_group}#{entity_id}`

In this series, we use a compound key that encodes both the entity group (A–Z) and a zero-padded entity ID:

```
A#000001
A#000002
...
B#000001
...
Z#005000
```

This gives us:
- **Prefix scans**: read all entities in group "A" by scanning rows with prefix `A#`
- **Point reads**: look up a single entity by exact key `A#000001`
- **Natural ordering**: entities within a group are sorted by ID

### Separator Choice

| Separator | Pros | Cons |
|-----------|------|------|
| `#` | Human-readable, common convention | Could appear in entity IDs |
| `\|` (pipe) | Visually distinct | Shell escaping issues |
| `\x00` (null byte) | Cannot appear in UTF-8 strings, so unambiguous | Not human-readable, complicates debugging |
| `:` (colon) | Common in Redis-style keys | Could conflict with timestamps |

````python
# Examine the key structure of our entity data in BigQuery
key_sample = bq.query(f'''
    SELECT
        entity_group,
        entity_id,
        CONCAT(entity_group, '#', LPAD(CAST(entity_id AS STRING), 6, '0')) AS row_key
    FROM `{BQ_PROJECT}.{BQ_DATASET}.{BQ_SOURCE_TABLE_A}`
    ORDER BY entity_group, entity_id
    LIMIT 10
''').to_dataframe()

print('Sample row keys from our entity data:')
print(key_sample.to_string(index=False))
````

````python
# Distribution of entities across groups
group_dist = bq.query(f'''
    SELECT
        entity_group,
        COUNT(*) AS entity_count,
        MIN(entity_id) AS min_id,
        MAX(entity_id) AS max_id,
        CONCAT(entity_group, '#', LPAD(CAST(MIN(entity_id) AS STRING), 6, '0')) AS first_key,
        CONCAT(entity_group, '#', LPAD(CAST(MAX(entity_id) AS STRING), 6, '0')) AS last_key
    FROM `{BQ_PROJECT}.{BQ_DATASET}.{BQ_SOURCE_TABLE_A}`
    GROUP BY entity_group
    ORDER BY entity_group
''').to_dataframe()

print(f'Entity distribution across {len(group_dist)} groups:')
print(f'Total entities: {group_dist["entity_count"].sum():,}')
print()
print(group_dist.to_string(index=False))
````

````python
# Key length analysis
# Our keys are 8 bytes: 1 (group) + 1 (#) + 6 (zero-padded ID)
example_keys = [
    'A#000001',    # 8 bytes - our format
    'A#1',          # 3 bytes - without padding
    'GROUP_A#000001',  # 14 bytes - verbose prefix
    'a1b2c3d4#A#000001',  # 18 bytes - hash prefix added
]

print('Key length comparison:')
print(f'{"Key":<30s} {"Bytes":>6s}  Notes')
print('-' * 70)
for key in example_keys:
    size = len(key.encode('utf-8'))
    notes = ''
    if key == 'A#000001':
        notes = '<-- our format (compact, sortable)'
    elif key == 'A#1':
        notes = 'no padding: A#1 sorts after A#10!'
    elif 'GROUP' in key:
        notes = 'verbose: wastes storage per row'
    elif 'a1b2' in key:
        notes = 'hash prefix: good distribution, longer key'
    print(f'{key:<30s} {size:>6d}  {notes}')

# Storage overhead at scale
total_entities = group_dist['entity_count'].sum()
compact_total = total_entities * 8  # our format
verbose_total = total_entities * 14  # verbose format
print(f'\nStorage overhead for {total_entities:,} entities:')
print(f'  Compact keys (8 bytes):  {compact_total:>12,} bytes ({compact_total / 1024 / 1024:.2f} MB)')
print(f'  Verbose keys (14 bytes): {verbose_total:>12,} bytes ({verbose_total / 1024 / 1024:.2f} MB)')
print(f'  Difference:              {verbose_total - compact_total:>12,} bytes')
````

**Key takeaway:** Our `{entity_group}#{entity_id}` format is compact (8 bytes), human-readable, and supports both prefix scans and point reads. The zero-padding on `entity_id` ensures correct lexicographic ordering: `A#000009` sorts before `A#000010`.

## Section 2: Key Distribution Visualization

Good key design distributes data evenly across the Bigtable cluster. Each Bigtable **tablet** (analogous to a shard) is responsible for a contiguous range of row keys. If keys are clustered in one area of the key space, one tablet handles a disproportionate amount of traffic.

Let's visualize how our compound keys distribute.

````python
# Visualize entity distribution across groups
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Bar chart: entities per group
ax1 = axes[0]
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(group_dist)))
ax1.bar(group_dist['entity_group'], group_dist['entity_count'], color=colors)
ax1.axhline(y=group_dist['entity_count'].mean(), color='red', linestyle='--',
            label=f'Mean: {group_dist["entity_count"].mean():,.0f}')
ax1.set_xlabel('Entity Group (Row Key Prefix)', fontweight='bold')
ax1.set_ylabel('Number of Entities', fontweight='bold')
ax1.set_title('Entities Per Group', fontsize=13, fontweight='bold')
ax1.legend()

# Pie chart showing proportional distribution
ax2 = axes[1]
# Group small slices for readability
counts = group_dist.set_index('entity_group')['entity_count']
ax2.pie(counts, labels=counts.index, autopct='%1.0f%%',
        textprops={'fontsize': 7}, pctdistance=0.85)
ax2.set_title('Key Space Distribution', fontsize=13, fontweight='bold')

plt.suptitle('Row Key Distribution Analysis', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Distribution statistics
cv = group_dist['entity_count'].std() / group_dist['entity_count'].mean()
print(f'Distribution statistics:')
print(f'  Groups:  {len(group_dist)}')
print(f'  Min:     {group_dist["entity_count"].min():,} entities')
print(f'  Max:     {group_dist["entity_count"].max():,} entities')
print(f'  Mean:    {group_dist["entity_count"].mean():,.0f} entities')
print(f'  Std:     {group_dist["entity_count"].std():,.0f}')
print(f'  CV:      {cv:.3f} (closer to 0 = more uniform)')
````

With our uniform entity groups (A–Z), each prefix represents roughly the same number of rows. This is the **ideal** scenario for Bigtable — no single tablet handles a disproportionate share of reads or writes.

In a real feature store, entity groups might represent categories (users, products, sessions) with very different cardinalities. That skew is where key design becomes critical.

## Section 3: Avoiding Hotspots — Anti-Patterns

A **hotspot** occurs when a small number of tablets handle a disproportionate amount of read or write traffic. The most common causes are row key designs that concentrate access in one region of the key space. For time-series workloads, see Google's guidance on [row key design for time series](https://cloud.google.com/bigtable/docs/schema-design-time-series).

### Anti-Pattern 1: Monotonically Increasing Keys

Using timestamps, auto-increment IDs, or sequential counters as keys (or key prefixes) sends all new writes to the **same tablet** — the one responsible for the latest key range. This creates a write hotspot that limits throughput to a single node.

```
BAD:                           GOOD:
2024-01-01T00:00:00#entity1    A#000001
2024-01-01T00:00:01#entity2    B#000042
2024-01-01T00:00:02#entity3    C#000007
2024-01-01T00:00:03#entity1    A#000099
     ^── all land on             ^── spread across
     same tablet               multiple tablets
```

````python
# Demonstrate: monotonically increasing keys vs our keys
# Simulate 1000 writes with timestamps vs entity keys

import datetime

# Anti-pattern: timestamp-prefixed keys
base_time = datetime.datetime(2024, 1, 1)
timestamp_keys = []
for i in range(1000):
    ts = base_time + datetime.timedelta(seconds=i)
    entity = f'entity_{i % 100:04d}'
    key = f'{ts.isoformat()}#{entity}'
    timestamp_keys.append(key)

# Good pattern: our entity group keys
entity_keys = []
groups = [chr(ord('A') + (i % 26)) for i in range(1000)]
for i in range(1000):
    group = chr(ord('A') + (i % 26))
    eid = (i // 26) + 1
    key = f'{group}#{eid:06d}'
    entity_keys.append(key)

# Analyze: which tablet (first 4 chars as proxy) each key maps to
def key_prefix_distribution(keys, prefix_len=4):
    """Count how many keys share the same prefix."""
    counts = defaultdict(int)
    for k in keys:
        counts[k[:prefix_len]] += 1
    return counts

ts_dist = key_prefix_distribution(timestamp_keys, prefix_len=13)  # date portion
ent_dist = key_prefix_distribution(entity_keys, prefix_len=1)  # group letter

print('=== Anti-Pattern: Timestamp-Prefixed Keys ===')
print(f'  Unique prefixes (simulated tablets): {len(ts_dist)}')
print(f'  Max keys per prefix: {max(ts_dist.values())}')
print(f'  All writes land sequentially on the SAME tablet region')
print(f'  Sample keys: {timestamp_keys[0]}')
print(f'               {timestamp_keys[1]}')
print(f'               {timestamp_keys[999]}')
print()
print('=== Good Pattern: Entity Group Keys ===')
print(f'  Unique prefixes (simulated tablets): {len(ent_dist)}')
print(f'  Keys per prefix: {dict(sorted(ent_dist.items())[:5])} ...')
print(f'  Writes spread across {len(ent_dist)} tablet regions')
print(f'  Sample keys: {entity_keys[0]}')
print(f'               {entity_keys[1]}')
print(f'               {entity_keys[999]}')
````

````python
# Visualize the hotspot
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Timestamp keys: show key positions in sorted order
ax1 = axes[0]
# Use first byte as proxy for position in key space
ts_positions = [ord(k[0]) for k in sorted(timestamp_keys)]
ax1.hist(ts_positions, bins=50, color='#E53935', alpha=0.8, edgecolor='white')
ax1.set_xlabel('First byte of key (key space position)', fontweight='bold')
ax1.set_ylabel('Number of keys', fontweight='bold')
ax1.set_title('Timestamp Keys: ALL in one region', fontsize=12, fontweight='bold', color='red')
ax1.annotate('HOTSPOT!', xy=(0.5, 0.9), xycoords='axes fraction',
             fontsize=20, color='red', fontweight='bold', ha='center')

# Entity keys: show key positions
ax2 = axes[1]
ent_positions = [ord(k[0]) for k in sorted(entity_keys)]
ax2.hist(ent_positions, bins=26, color='#1976D2', alpha=0.8, edgecolor='white')
ax2.set_xlabel('First byte of key (key space position)', fontweight='bold')
ax2.set_ylabel('Number of keys', fontweight='bold')
ax2.set_title('Entity Group Keys: Evenly Distributed', fontsize=12, fontweight='bold', color='blue')
ax2.annotate('Uniform', xy=(0.5, 0.9), xycoords='axes fraction',
             fontsize=20, color='green', fontweight='bold', ha='center')

plt.suptitle('Write Distribution: Timestamp vs Entity Group Keys',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

### Anti-Pattern 2: Concentrated Prefixes

If your data has a skewed distribution — say 90% of entities belong to group "A" — then one tablet handles 90% of traffic even though the key format is correct.

This is a **data skew** problem, not a key format problem. The fix is to add a hash prefix or salt to spread the dominant group across multiple tablets.

````python
# Simulate a skewed distribution
total = group_dist['entity_count'].sum()

# Create a skewed version: 90% in group A
skewed = pd.DataFrame({
    'entity_group': list('ABCDEFGHIJKLMNOPQRSTUVWXYZ'),
    'skewed_count': [int(total * 0.90)] + [int(total * 0.10 / 25)] * 25
})

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Uniform (our data)
axes[0].bar(group_dist['entity_group'], group_dist['entity_count'], color='#1976D2')
axes[0].set_title('Our Data: Uniform Distribution', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Entity Group')
axes[0].set_ylabel('Entity Count')

# Skewed
colors_skewed = ['#E53935'] + ['#1976D2'] * 25
axes[1].bar(skewed['entity_group'], skewed['skewed_count'], color=colors_skewed)
axes[1].set_title('Skewed: 90% in Group A (Hotspot!)', fontsize=12, fontweight='bold', color='red')
axes[1].set_xlabel('Entity Group')
axes[1].set_ylabel('Entity Count')

plt.suptitle('Key Prefix Distribution: Uniform vs Skewed',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print('If 90% of reads target group A, that single tablet handles 90% of traffic.')
print('The fix: add a hash prefix to spread group A across multiple tablets.')
````

### Anti-Pattern 3: Hash-Only Keys

Using a hash of the entity ID as the entire key provides **perfect distribution** but destroys the ability to do prefix scans or range reads. Every read becomes a point read.

This is the opposite extreme — maximum distribution but zero locality.

````python
# Compare: original key vs hash-only key
sample_entities = [
    ('A', 1), ('A', 2), ('A', 3), ('A', 4), ('A', 5),
    ('B', 1), ('B', 2), ('B', 3),
]

print(f'{"Original Key":<15s} {"Hash-Only Key":<20s} {"Sortable?":>10s} {"Scannable?":>12s}')
print('-' * 60)
for group, eid in sample_entities:
    original = f'{group}#{eid:06d}'
    hash_key = hashlib.sha256(f'{group}_{eid}'.encode()).hexdigest()[:16]
    print(f'{original:<15s} {hash_key:<20s} {"Yes":>10s} {"No":>12s}')

print()
print('With original keys:  scan A# returns A#000001, A#000002, A#000003, A#000004, A#000005')
print('With hash-only keys: no prefix scan possible - must know every exact key')
````

## Section 4: Key Design Patterns for Feature Stores

There are several well-established key design patterns, each with different tradeoffs. The right choice depends on your access patterns.

### Pattern 1: Namespace Prefix (Our Approach)

```
{entity_group}#{entity_id}
```

- Supports prefix scans (all entities in a group)
- Natural grouping for feature retrieval
- Risk: hotspots if one group dominates

### Pattern 2: Hash-Prefix for Distribution

```
{hash(entity_id)[:4]}#{entity_group}#{entity_id}
```

- Even distribution regardless of group size
- Loses group locality (no efficient group scan)
- Good for point reads when group scans are rare

### Pattern 3: Reverse Domain Style

```
{entity_id_reversed}#{entity_group}
```

- Distributes sequential IDs (000001 → 100000, 000002 → 200000)
- Useful when entity IDs are auto-incremented
- Harder to read and debug

### Pattern 4: Salted Keys

```
{salt}#{entity_group}#{entity_id}   where salt = entity_id mod N
```

- Spreads writes across N buckets
- Requires N parallel scans to read a full group
- N is tunable (more salt buckets = better distribution, worse scan performance)

````python
# Generate each key format from our data
def generate_key_formats(group, entity_id):
    """Generate all four key patterns for a given entity."""
    eid_str = f'{entity_id:06d}'
    entity_str = f'{group}_{eid_str}'

    # Pattern 1: Namespace prefix (our approach)
    p1 = f'{group}#{eid_str}'

    # Pattern 2: Hash-prefix for distribution
    hash_prefix = hashlib.md5(entity_str.encode()).hexdigest()[:4]
    p2 = f'{hash_prefix}#{group}#{eid_str}'

    # Pattern 3: Reverse domain style
    reversed_id = eid_str[::-1]
    p3 = f'{reversed_id}#{group}'

    # Pattern 4: Salted keys (8 salt buckets)
    salt = entity_id % 8
    p4 = f'{salt:02d}#{group}#{eid_str}'

    return p1, p2, p3, p4

# Show key formats for sample entities
print(f'{"Entity":<12s} {"P1: Namespace":<15s} {"P2: Hash-Prefix":<25s} {"P3: Reversed":<15s} {"P4: Salted":<20s}')
print('=' * 90)

samples = [('A', 1), ('A', 2), ('A', 100), ('B', 1), ('B', 50), ('Z', 5000)]
for group, eid in samples:
    p1, p2, p3, p4 = generate_key_formats(group, eid)
    print(f'{group}#{eid:<7d} {p1:<15s} {p2:<25s} {p3:<15s} {p4:<20s}')
````

````python
# Compare: which patterns support prefix scans?
comparison = pd.DataFrame([
    {'Pattern': 'P1: Namespace Prefix', 'Prefix Scans': 'Yes (by group)',
     'Point Reads': 'Yes', 'Distribution': 'Depends on group sizes',
     'Complexity': 'Low', 'Best For': 'Group-level serving'},
    {'Pattern': 'P2: Hash-Prefix', 'Prefix Scans': 'No',
     'Point Reads': 'Yes', 'Distribution': 'Excellent (uniform)',
     'Complexity': 'Medium', 'Best For': 'High-throughput point reads'},
    {'Pattern': 'P3: Reversed', 'Prefix Scans': 'No',
     'Point Reads': 'Yes', 'Distribution': 'Good (breaks sequences)',
     'Complexity': 'Medium', 'Best For': 'Sequential ID sources'},
    {'Pattern': 'P4: Salted', 'Prefix Scans': 'Partial (N scans)',
     'Point Reads': 'Yes', 'Distribution': 'Good (N buckets)',
     'Complexity': 'High', 'Best For': 'Balanced distribution + scans'},
])

print('Key Design Pattern Comparison:')
print(comparison.to_string(index=False))
````

````python
# Visualize distribution of each pattern
# Generate keys for all our entities and check first-byte distribution
sample_size = 1000
np.random.seed(42)
sample_groups = np.random.choice(list('ABCDEFGHIJKLMNOPQRSTUVWXYZ'), sample_size)
sample_ids = np.random.randint(1, 5001, sample_size)

patterns = {'P1: Namespace': [], 'P2: Hash-Prefix': [], 'P3: Reversed': [], 'P4: Salted': []}
for g, eid in zip(sample_groups, sample_ids):
    p1, p2, p3, p4 = generate_key_formats(g, eid)
    patterns['P1: Namespace'].append(ord(p1[0]))
    patterns['P2: Hash-Prefix'].append(ord(p2[0]))
    patterns['P3: Reversed'].append(ord(p3[0]))
    patterns['P4: Salted'].append(ord(p4[0]))

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
colors = ['#1976D2', '#E53935', '#43A047', '#FF9800']

for idx, (name, values) in enumerate(patterns.items()):
    ax = axes[idx // 2][idx % 2]
    ax.hist(values, bins=40, color=colors[idx], alpha=0.8, edgecolor='white')
    ax.set_title(name, fontsize=12, fontweight='bold')
    ax.set_xlabel('First byte value (position in key space)')
    ax.set_ylabel('Count')

plt.suptitle('Key Space Distribution by Pattern (1000 random entities)',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

**Interpretation:**

## Section 5: Prefix Scans and Row Filters

One of Bigtable's strengths is efficient **prefix scans** — [reading](https://cloud.google.com/bigtable/docs/reads) all rows whose key starts with a given prefix. Combined with [row filters](https://cloud.google.com/bigtable/docs/using-filters), you can selectively read specific columns from a subset of rows.

### Key operations:

| Operation | Use Case | API |
|-----------|----------|-----|
| **Point read** | Fetch one entity by exact key | `table.read_row(key)` |
| **Prefix scan** | Fetch all entities in a group | `table.read_rows(row_set=..., filter_=...)` with `RowRange` |
| **Multi-row read** | Fetch specific entities by key list | `table.read_rows(row_set=...)` with `RowSet` |
| **Filtered read** | Fetch specific columns only | `ColumnQualifierRegexFilter` or `FamilyNameRegexFilter` |

````python
# Create a demo table for key design experiments
demo_table = instance.table(BT_TABLE_KEY_DESIGN)

if demo_table.exists():
    demo_table.delete()
    print(f'Deleted existing table: {BT_TABLE_KEY_DESIGN}')

demo_table.create(column_families={
    'features': cf.MaxVersionsGCRule(1),
})
print(f'Created table: {BT_TABLE_KEY_DESIGN}')
print(f'  Column family: features (max 1 version)')
````

````python
# Write sample data with our compound keys
# Use a subset of our BQ data to populate the demo table
sample_data = bq.query(f'''
    SELECT
        entity_group,
        entity_id,
        feature_int_1,
        feature_float_1,
        feature_int_2,
        feature_float_2
    FROM `{BQ_PROJECT}.{BQ_DATASET}.{BQ_SOURCE_TABLE_A}`
    WHERE entity_group IN ('A', 'B', 'C')
    ORDER BY entity_group, entity_id
    LIMIT 300
''').to_dataframe()

print(f'Fetched {len(sample_data)} rows from BigQuery')
print(f'Groups: {sorted(sample_data["entity_group"].unique())}')

# Write to Bigtable in batches
rows_to_write = []
for _, row in sample_data.iterrows():
    key = f'{row["entity_group"]}#{row["entity_id"]:06d}'
    bt_row = demo_table.direct_row(key)
    bt_row.set_cell('features', 'feature_int_1', str(row['feature_int_1']).encode('utf-8'))
    bt_row.set_cell('features', 'feature_float_1', str(row['feature_float_1']).encode('utf-8'))
    bt_row.set_cell('features', 'feature_int_2', str(row['feature_int_2']).encode('utf-8'))
    bt_row.set_cell('features', 'feature_float_2', str(row['feature_float_2']).encode('utf-8'))
    rows_to_write.append(bt_row)

# Mutate in batches of 100
batch_size = 100
for i in range(0, len(rows_to_write), batch_size):
    batch = rows_to_write[i:i + batch_size]
    statuses = demo_table.mutate_rows(batch)
    errors = [s for s in statuses if s.code != 0]
    if errors:
        print(f'  Batch {i // batch_size}: {len(errors)} errors')
    else:
        print(f'  Batch {i // batch_size}: wrote {len(batch)} rows')

print(f'\nTotal rows written: {len(rows_to_write)}')
````

### Point Read: Single Entity

````python
# Point read: fetch a single entity by exact key
target_key = 'A#000001'

start = time.time()
row = demo_table.read_row(target_key.encode('utf-8'))
elapsed = time.time() - start

if row:
    print(f'Point read for key: {target_key} ({elapsed * 1000:.1f} ms)')
    for family, columns in row.cells.items():
        for qualifier, versions in columns.items():
            value = versions[0].value.decode('utf-8')
            print(f'  {family}:{qualifier.decode("utf-8")} = {value}')
else:
    print(f'Key not found: {target_key}')
````

### Prefix Scan: All Entities in a Group

````python
# Prefix scan: read all rows with key prefix "A#"
prefix = 'A#'

# Create a RowRange for prefix scan
# Start key: prefix (inclusive)
# End key: prefix with last byte incremented (exclusive)
start_key = prefix.encode('utf-8')
end_key = prefix[:-1].encode('utf-8') + bytes([prefix[-1:].encode('utf-8')[0] + 1])

row_range = row_set_mod.RowRange(start_key=start_key, end_key=end_key)
row_set = row_set_mod.RowSet()
row_set.add_row_range(row_range)

start = time.time()
rows = list(demo_table.read_rows(row_set=row_set))
elapsed = time.time() - start

print(f'Prefix scan for "{prefix}*": {len(rows)} rows in {elapsed * 1000:.1f} ms')
print(f'First key: {rows[0].row_key.decode("utf-8") if rows else "none"}')
print(f'Last key:  {rows[-1].row_key.decode("utf-8") if rows else "none"}')
````

### Filtered Read: Specific Columns from a Group

````python
# Filtered scan: read only feature_int_1 and feature_float_1 for group "A"
# Use a ColumnQualifierRegexFilter to select specific columns
col_filter = row_filters.ColumnQualifierRegexFilter(b'feature_int_1|feature_float_1')

start = time.time()
filtered_rows = list(demo_table.read_rows(row_set=row_set, filter_=col_filter))
elapsed = time.time() - start

print(f'Filtered scan (2 columns) for group "A": {len(filtered_rows)} rows in {elapsed * 1000:.1f} ms')
print()

# Show first 5 rows
print(f'{"Row Key":<12s} {"feature_int_1":>15s} {"feature_float_1":>17s}')
print('-' * 46)
for row in filtered_rows[:5]:
    key = row.row_key.decode('utf-8')
    vals = {}
    for family, columns in row.cells.items():
        for qualifier, versions in columns.items():
            vals[qualifier.decode('utf-8')] = versions[0].value.decode('utf-8')
    print(f'{key:<12s} {vals.get("feature_int_1", "N/A"):>15s} {vals.get("feature_float_1", "N/A"):>17s}')
print(f'... ({len(filtered_rows)} total rows)')
````

### Multi-Row Read: Specific Entities by Key

````python
# Multi-row read: fetch specific entities across groups
target_keys = ['A#000001', 'A#000050', 'B#000001', 'C#000010']

row_set_specific = row_set_mod.RowSet()
for key in target_keys:
    row_set_specific.add_row_key(key.encode('utf-8'))

start = time.time()
multi_rows = list(demo_table.read_rows(row_set=row_set_specific))
elapsed = time.time() - start

print(f'Multi-row read for {len(target_keys)} specific keys: {len(multi_rows)} found in {elapsed * 1000:.1f} ms')
print()
for row in multi_rows:
    key = row.row_key.decode('utf-8')
    col_count = sum(len(cols) for cols in row.cells.values())
    print(f'  {key}: {col_count} columns')
````

````python
# Latency comparison: point read vs prefix scan vs filtered scan
def measure_latency(operation_fn, label, n=10):
    """Run an operation n times and report latency statistics."""
    latencies = []
    for _ in range(n):
        start = time.time()
        result = operation_fn()
        latencies.append((time.time() - start) * 1000)  # ms
    return {
        'operation': label,
        'p50_ms': round(np.percentile(latencies, 50), 2),
        'p95_ms': round(np.percentile(latencies, 95), 2),
        'p99_ms': round(np.percentile(latencies, 99), 2),
        'mean_ms': round(np.mean(latencies), 2),
    }

# Point read
point_stats = measure_latency(
    lambda: demo_table.read_row('A#000001'.encode('utf-8')),
    'Point read (1 row)'
)

# Prefix scan (group A)
scan_stats = measure_latency(
    lambda: list(demo_table.read_rows(row_set=row_set)),
    f'Prefix scan (group A)'
)

# Filtered scan
filtered_stats = measure_latency(
    lambda: list(demo_table.read_rows(row_set=row_set, filter_=col_filter)),
    'Filtered scan (2 cols, group A)'
)

# Multi-row read
multi_stats = measure_latency(
    lambda: list(demo_table.read_rows(row_set=row_set_specific)),
    f'Multi-row read ({len(target_keys)} keys)'
)

latency_df = pd.DataFrame([point_stats, scan_stats, filtered_stats, multi_stats])
print('Latency comparison (10 iterations each):')
print(latency_df.to_string(index=False))
````

**Key observations:**

- **Point reads** are the fastest operation — Bigtable can locate a single row by key in microseconds.
- **Prefix scans** return many rows but leverage the sorted key order for efficient sequential reads.
- **Filtered scans** may not be faster than unfiltered scans on small datasets, because the filter overhead can exceed the savings from reading fewer columns. On large rows (hundreds of columns), column filters dramatically reduce data transfer.
- **Multi-row reads** are efficient for fetching scattered keys — Bigtable parallelizes the lookups internally.

## Section 6: Column Family Strategy

[Column families](https://cloud.google.com/bigtable/docs/schema-design#column-families) in Bigtable are **physical storage units**. Each family is stored in its own set of files (SSTables) on disk. This has important implications:

| Property | Description |
|----------|-------------|
| **Physical separation** | Each family is stored independently. Reading from one family does not touch data in other families. |
| **GC policy** | Each family has its own garbage collection policy (max versions, max age, or a combination). |
| **Access patterns** | Group columns that are read together into the same family for I/O efficiency. |
| **Limit** | Maximum 100 column families per table (Bigtable hard limit). |

### Column Family Design for Feature Stores

A well-designed feature store uses multiple column families to separate different access patterns:

| Family | Purpose | GC Policy | Access Pattern |
|--------|---------|-----------|----------------|
| `features` | Main feature values | Keep 1 version | Bulk reads during inference |
| `metadata` | Schema info, version, timestamps | Keep 1 version | Occasional reads for schema discovery |
| `streaming` | Real-time features (clicks, counts) | Keep 5 versions | Frequent writes, recent value reads |

````python
# Create a multi-family table to demonstrate column family separation
multi_cf_table = instance.table(BT_TABLE_MULTI_CF)

if multi_cf_table.exists():
    multi_cf_table.delete()
    print(f'Deleted existing table: {BT_TABLE_MULTI_CF}')

multi_cf_table.create(column_families={
    'features': cf.MaxVersionsGCRule(1),
    'metadata': cf.MaxVersionsGCRule(1),
    'streaming': cf.MaxVersionsGCRule(5),
})

print(f'Created table: {BT_TABLE_MULTI_CF}')
print(f'Column families:')
for name, family in multi_cf_table.list_column_families().items():
    print(f'  {name}: GC rule = {family.gc_rule}')
````

````python
# Write data to different families
# This demonstrates the separation of concerns

import datetime

sample_keys_multi = ['A#000001', 'A#000002', 'B#000001']

rows_multi = []
for key in sample_keys_multi:
    bt_row = multi_cf_table.direct_row(key)

    # Features family: the main feature values
    bt_row.set_cell('features', 'feature_int_1', b'42')
    bt_row.set_cell('features', 'feature_float_1', b'3.14')
    bt_row.set_cell('features', 'feature_int_2', b'100')

    # Metadata family: schema and version info
    bt_row.set_cell('metadata', 'schema_version', b'v2')
    bt_row.set_cell('metadata', 'last_updated', 
                    datetime.datetime.now().isoformat().encode('utf-8'))
    bt_row.set_cell('metadata', 'source', b'bigquery_export')

    # Streaming family: real-time features (simulated multiple versions)
    bt_row.set_cell('streaming', 'click_count', b'15')
    bt_row.set_cell('streaming', 'last_action', b'purchase')

    rows_multi.append(bt_row)

statuses = multi_cf_table.mutate_rows(rows_multi)
errors = [s for s in statuses if s.code != 0]
print(f'Wrote {len(rows_multi)} rows to {BT_TABLE_MULTI_CF} ({len(errors)} errors)')
````

````python
# Read selectively from different families

# Read ONLY the features family (most common access pattern during inference)
features_filter = row_filters.FamilyNameRegexFilter(b'features')

row = multi_cf_table.read_row('A#000001'.encode('utf-8'), filter_=features_filter)
print('=== Read only "features" family ===')
for family, columns in row.cells.items():
    for qualifier, versions in columns.items():
        print(f'  {family}:{qualifier.decode("utf-8")} = {versions[0].value.decode("utf-8")}')

print()

# Read ONLY the metadata family (for schema discovery)
metadata_filter = row_filters.FamilyNameRegexFilter(b'metadata')

row = multi_cf_table.read_row('A#000001'.encode('utf-8'), filter_=metadata_filter)
print('=== Read only "metadata" family ===')
for family, columns in row.cells.items():
    for qualifier, versions in columns.items():
        print(f'  {family}:{qualifier.decode("utf-8")} = {versions[0].value.decode("utf-8")}')

print()

# Read ALL families (full row)
row = multi_cf_table.read_row('A#000001'.encode('utf-8'))
print('=== Read all families (full row) ===')
for family, columns in sorted(row.cells.items()):
    for qualifier, versions in columns.items():
        print(f'  {family}:{qualifier.decode("utf-8")} = {versions[0].value.decode("utf-8")}')
````

````python
# Demonstrate GC policy difference: streaming family keeps 5 versions
# Write multiple versions to the streaming family
key = 'A#000001'

for i in range(7):
    bt_row = multi_cf_table.direct_row(key)
    bt_row.set_cell(
        'streaming', 'click_count',
        str(10 + i).encode('utf-8'),
        # Each write uses a different timestamp so they are treated as separate versions
        timestamp=datetime.datetime(2024, 1, 1, 0, 0, i)
    )
    bt_row.commit()

print(f'Wrote 7 versions of click_count to streaming family for {key}')

# Read with CellsColumnLimitFilter to see all retained versions
versions_filter = row_filters.CellsColumnLimitFilter(10)  # up to 10 versions
streaming_filter = row_filters.RowFilterChain(filters=[
    row_filters.FamilyNameRegexFilter(b'streaming'),
    row_filters.ColumnQualifierRegexFilter(b'click_count'),
    versions_filter,
])

row = multi_cf_table.read_row(key.encode('utf-8'), filter_=streaming_filter)
if row:
    for family, columns in row.cells.items():
        for qualifier, versions in columns.items():
            print(f'\nVersions of {family}:{qualifier.decode("utf-8")} (GC keeps max 5):')
            for v in versions:
                print(f'  timestamp={v.timestamp}  value={v.value.decode("utf-8")}')
            print(f'\nRetained: {len(versions)} versions (GC may not have run yet; max is 5)')
````

**Column family design guidance:**

1. **Group by access pattern.** If features are always read together during inference, put them in the same family. If metadata is only read during schema discovery, separate it.
2. **Use GC policies wisely.** Batch features need only 1 version (latest). Streaming features may benefit from keeping recent history (5–10 versions) for time-windowed aggregations.
3. **Do not create a family per column.** This wastes the 100-family limit and creates unnecessary SSTable files. Use column qualifiers within a family instead.
4. **Name families descriptively.** Short but meaningful: `features`, `metadata`, `streaming`, `embeddings`.

## Section 7: Single Table vs Multi-Table

A fundamental design decision for feature stores is whether to store all feature groups in a **single table** or use **separate tables** per feature group or entity type.

### Single Table

All feature groups in one table, differentiated by column families or column qualifier prefixes.

```
Table: feature-store
┌────────────┬───────────────────┬───────────────────┬───────────────────┐
│ Row Key    │ cf:user_features  │ cf:product_features │ cf:session_features │
├────────────┼───────────────────┼───────────────────┼───────────────────┤
│ A#000001   │ age=25, clicks=10 │ (empty)             │ duration=300s       │
│ B#000001   │ age=30            │ price=49.99         │ (empty)             │
└────────────┴───────────────────┴───────────────────┴───────────────────┘
```

### Multi-Table

Separate tables per feature group or entity type.

```
Table: user-features          Table: product-features
┌───────────┬───────────────┐   ┌───────────┬───────────────┐
│ Row Key   │ cf:features     │   │ Row Key   │ cf:features     │
├───────────┼───────────────┤   ├───────────┼───────────────┤
│ user_001  │ age=25          │   │ prod_001  │ price=49.99     │
│ user_002  │ age=30          │   │ prod_002  │ price=29.99     │
└───────────┴───────────────┘   └───────────┴───────────────┘
```

````python
# Decision matrix: Single Table vs Multi-Table
decision = pd.DataFrame([
    {'Criterion': 'Management complexity',
     'Single Table': 'Lower (one table to manage)',
     'Multi-Table': 'Higher (N tables, N configs)'},
    {'Criterion': 'Cross-group reads',
     'Single Table': 'Atomic (one read_row call)',
     'Multi-Table': 'Multiple calls required'},
    {'Criterion': 'GC policy granularity',
     'Single Table': 'Per column family (up to 100)',
     'Multi-Table': 'Per table + per family'},
    {'Criterion': 'Schema evolution',
     'Single Table': 'Add columns freely (sparse)',
     'Multi-Table': 'Independent schemas per table'},
    {'Criterion': 'Scaling independence',
     'Single Table': 'Shared (all groups scale together)',
     'Multi-Table': 'Independent (per-table scaling)'},
    {'Criterion': 'Column family limit',
     'Single Table': 'Shared 100-family limit',
     'Multi-Table': '100 families per table'},
    {'Criterion': 'Access control',
     'Single Table': 'Table-level or family-level IAM',
     'Multi-Table': 'Independent table-level IAM'},
    {'Criterion': 'Backup & restore',
     'Single Table': 'All-or-nothing per table',
     'Multi-Table': 'Independent backup per group'},
])

print('Single Table vs Multi-Table Decision Matrix:')
print(decision.to_string(index=False))
````

````python
# Demonstrate: single-table approach with multiple column families
# Our multi-cf table already shows this pattern

# Read user A#000001 with all feature groups in one call
row = multi_cf_table.read_row('A#000001'.encode('utf-8'))

print('Single-table read for A#000001 (all families):')
for family in sorted(row.cells.keys()):
    columns = row.cells[family]
    col_list = ', '.join(f'{q.decode("utf-8")}={v[0].value.decode("utf-8")}'
                        for q, v in columns.items())
    print(f'  [{family}] {col_list}')

print()
print('Advantage: all feature groups retrieved in ONE read_row() call.')
print('With multi-table: this would require 3 separate read_row() calls to 3 tables.')
````

### Guidance: When to Split

Start with a **single table** and split into multiple tables when you encounter one of these:

1. **Access pattern conflicts.** If batch features need hourly GC but streaming features need per-minute writes with 5-version retention, and you are running out of column families.
2. **Scaling independence.** If user features need 10 nodes but product features need 2 nodes, separate tables let you size each independently.
3. **IAM boundaries.** If different teams own different feature groups and need independent access control.
4. **Backup/restore granularity.** If you need to restore one feature group without affecting others.

For most feature stores — especially those under 50 feature groups — a single table with well-designed column families is the simpler, more efficient choice. Use [app profiles](https://cloud.google.com/bigtable/docs/app-profiles) to route different traffic types (serving vs batch) to appropriate clusters.

### What to Look For in Key Visualizer

The Key Visualizer shows a heatmap where:
- **X-axis** = time
- **Y-axis** = row key range (sorted lexicographically)
- **Color intensity** = metric value (reads/sec, writes/sec, storage, etc.)

#### Healthy Patterns

| Pattern | What it looks like | What it means |
|---------|-------------------|---------------|
| **Even stripe** | Uniform color across the full Y-axis | Load is evenly distributed across all keys |
| **Periodic bands** | Alternating light/dark horizontal bands | Regular batch processing on different key ranges |
| **Gradual gradient** | Smooth color transition from top to bottom | Slight natural skew, but no single hotspot |

#### Problem Patterns

| Pattern | What it looks like | What it means |
|---------|-------------------|---------------|
| **Single bright row** | One thin horizontal line much brighter than the rest | Hotspot: one row key getting massive traffic |
| **Bright bottom edge** | Intense color only at the bottom of the heatmap | Monotonically increasing keys (all writes to latest key range) |
| **Bright top cluster** | Bright band at the top of the key space | One prefix (e.g., "A#") handles most traffic |
| **Diagonal stripe** | Bright line moving from bottom-left to top-right | Sequential scan repeatedly iterating through the key space |

### Metrics Available

Key Visualizer supports multiple metrics:

- **Ops (operations/sec)**: Read and write operations per key range
- **Bytes read/written**: Data volume per key range
- **Average rows per read**: How many rows are being returned per read operation
- **CPU usage**: Processing time per key range

For feature store monitoring, **Ops** and **Bytes read** are the most actionable metrics.

**Documentation:** [Bigtable Key Visualizer overview](https://cloud.google.com/bigtable/docs/keyvis-overview)

````python
# Simulate what Key Visualizer would show for our key distribution
# This is a conceptual visualization since the actual tool runs in Cloud Console

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Simulate heatmaps for different key patterns
time_steps = 60  # 60 time intervals
key_ranges = 26  # A through Z

# Pattern 1: Our uniform distribution (healthy)
np.random.seed(42)
uniform_heat = np.random.uniform(0.4, 0.6, (key_ranges, time_steps))
axes[0].imshow(uniform_heat, aspect='auto', cmap='YlOrRd', vmin=0, vmax=1)
axes[0].set_title('Healthy: Uniform Distribution', fontweight='bold', color='green')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Key Range (A\u2192Z)')
axes[0].set_yticks(range(0, 26, 5))
axes[0].set_yticklabels([chr(ord('A') + i) for i in range(0, 26, 5)])

# Pattern 2: Hotspot (one group dominates)
hotspot_heat = np.random.uniform(0.05, 0.15, (key_ranges, time_steps))
hotspot_heat[0, :] = np.random.uniform(0.85, 1.0, time_steps)  # Group A is hot
axes[1].imshow(hotspot_heat, aspect='auto', cmap='YlOrRd', vmin=0, vmax=1)
axes[1].set_title('Hotspot: Group A Dominates', fontweight='bold', color='red')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Key Range (A\u2192Z)')
axes[1].set_yticks(range(0, 26, 5))
axes[1].set_yticklabels([chr(ord('A') + i) for i in range(0, 26, 5)])

# Pattern 3: Monotonic writes (all writes to one growing region)
monotonic_heat = np.zeros((key_ranges, time_steps))
for t in range(time_steps):
    # Writes concentrate at an advancing position
    pos = int(t * key_ranges / time_steps)
    if pos < key_ranges:
        monotonic_heat[pos, t] = np.random.uniform(0.8, 1.0)
        if pos > 0:
            monotonic_heat[pos - 1, t] = np.random.uniform(0.2, 0.4)
axes[2].imshow(monotonic_heat, aspect='auto', cmap='YlOrRd', vmin=0, vmax=1)
axes[2].set_title('Monotonic: Sequential Writes', fontweight='bold', color='orange')
axes[2].set_xlabel('Time')
axes[2].set_ylabel('Key Range')
axes[2].set_yticks(range(0, 26, 5))
axes[2].set_yticklabels([chr(ord('A') + i) for i in range(0, 26, 5)])

plt.suptitle('Simulated Key Visualizer Heatmaps', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print('Left:   Even color = good. Load distributed across all key ranges.')
print('Center: Bright top row = bad. Group A handles 90%+ of traffic.')
print('Right:  Diagonal line = bad. Writes always go to the current/latest key range.')
````

## Summary

| Topic | Key Takeaway |
|-------|-------------|
| **Row key anatomy** | Keys are arbitrary bytes, sorted lexicographically. Our `{group}#{id}` format is compact and supports both prefix scans and point reads. |
| **Key distribution** | Uniform distribution across prefixes prevents hotspots. Visualize distribution before deploying. |
| **Hotspot anti-patterns** | Monotonically increasing keys, skewed prefixes, and hash-only keys each cause different problems. |
| **Key design patterns** | Namespace prefix, hash-prefix, reversed, and salted keys each trade off distribution vs scan capability. |
| **Prefix scans & filters** | RowRange for prefix scans, RowSet for multi-key reads, column/family filters for selective reads. |
| **Column families** | Physical storage units with independent GC policies. Separate by access pattern, not by column. |
| **Single vs multi-table** | Start with single table. Split when you hit access pattern conflicts, scaling needs, or IAM boundaries. |
| **Key Visualizer** | The primary diagnostic tool for hotspots. Look for uneven color intensity across the key space. |
