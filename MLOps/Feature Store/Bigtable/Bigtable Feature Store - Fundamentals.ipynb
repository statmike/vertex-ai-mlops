# Bigtable Feature Store — Fundamentals

BigQuery is the analytical powerhouse for feature engineering — window functions, joins, ML functions — but it's a batch system. When a customer taps "pay" or a sensor fires an alert, you need features in milliseconds, not seconds. [Cloud Bigtable](https://cloud.google.com/bigtable/docs/overview) provides single-digit millisecond reads by row key, making it the ideal online serving layer for pre-computed features.

This notebook shows the complete end-to-end flow: compute features in BigQuery, [export to Bigtable](https://cloud.google.com/bigquery/docs/export-to-bigtable), and serve them at production speed. By the end, you'll see the latency difference measured and visualized.

**Prerequisites:** Run [Notebook 0: Environment](./Bigtable%20Feature%20Store%20-%20Environment.ipynb) first to create the BigQuery data and Bigtable instance.

**What you'll learn:**
- Why Bigtable for feature serving (the latency problem, quantified)
- EXPORT DATA to write BQ query results directly to Bigtable
- Self-describing feature tables with schema metadata rows
- Reading features: single entity and batch lookups
- The latency shootout: BigQuery seconds vs Bigtable milliseconds

## 1. The Latency Problem

Analytical queries in BigQuery take seconds because every query must schedule a job, scan tables, compute aggregations, and return results. That's the price of a full SQL engine — it's optimized for throughput over latency.

For real-time serving — a checkout flow, a fraud decision, a recommendation — you need features in milliseconds. Let's measure the gap.

```python
TEST_ENTITY_GROUP = 'A'
TEST_ENTITY_ID = '000001'

start = time.perf_counter()
query = f"""
SELECT * FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
WHERE entity_group = '{TEST_ENTITY_GROUP}' AND entity_id = '{TEST_ENTITY_ID}'
"""
result = bq.query(query).to_dataframe()
bq_latency = time.perf_counter() - start

print(f'BigQuery latency: {bq_latency:.3f}s ({bq_latency*1000:.0f}ms)')
print(f'\nRows returned: {len(result)}')
print(f'Columns: {list(result.columns)}')
print(f'\nFeature values:')
print(result.T.to_string(header=False))
```

### Interpreting the Latency

The BigQuery query takes seconds — even for a single row lookup with a WHERE clause. That's because BigQuery is a distributed analytical engine: every query goes through job scheduling, metadata resolution, slot allocation, table scanning, and result serialization.

For a checkout flow or real-time scoring pipeline, seconds of latency is unacceptable. The ML model itself runs in microseconds — the bottleneck is **feature retrieval**. Pre-computing features and storing them in a key-value store eliminates the query overhead entirely.

---
## 2. EXPORT DATA Setup

[EXPORT DATA](https://cloud.google.com/bigquery/docs/reference/standard-sql/other-statements#export_data_statement) with `format = 'CLOUD_BIGTABLE'` is BigQuery's native mechanism for writing query results directly to Bigtable. It requires two pieces of infrastructure:

1. **Enterprise [reservation](https://cloud.google.com/bigquery/docs/reservations-intro)** — EXPORT DATA to Bigtable requires [BigQuery Enterprise edition](https://cloud.google.com/bigquery/docs/editions-intro). We create a reservation with 100 slots and a QUERY-type assignment so the export runs as a regular query job. Slots are billed per-second at on-demand Enterprise rates.

2. **Bigtable [app profile](https://cloud.google.com/bigtable/docs/app-profiles)** — the export must target an app profile configured with single-cluster routing and LOW request priority. This prevents the export from competing with latency-sensitive serving reads.

> **Cost note:** We create the reservation, run the export, and delete it immediately — total cost is typically under $0.10 for a small export like ours.

### Create Enterprise Reservation

```python
# Create Enterprise reservation for EXPORT DATA
# No capacity commitment needed — on-demand edition billing.
# Slots are billed per-second while the reservation exists.
parent = f'projects/{PROJECT_ID}/locations/{REGION}'

# Step 1: Create reservation with Enterprise edition
try:
    reservation = reservation_client.create_reservation(
        parent=parent,
        reservation_id=RESERVATION_ID,
        reservation=bigquery_reservation_v1.Reservation(
            slot_capacity=RESERVATION_SLOTS,
            edition=bigquery_reservation_v1.Edition.ENTERPRISE,
        ),
    )
    print(f'Created reservation: {reservation.name}')
    print(f'  Slots:   {reservation.slot_capacity}')
    print(f'  Edition: ENTERPRISE')
except AlreadyExists:
    reservation = reservation_client.get_reservation(
        name=f'{parent}/reservations/{RESERVATION_ID}'
    )
    print(f'Reservation exists: {reservation.name}')
```

```python
# Step 2: QUERY assignment — routes query jobs through the reservation
try:
    assignment = reservation_client.create_assignment(
        parent=reservation.name,
        assignment=bigquery_reservation_v1.Assignment(
            job_type=bigquery_reservation_v1.Assignment.JobType.QUERY,
            assignee=f'projects/{PROJECT_ID}',
        ),
    )
    print(f'Created assignment: {assignment.name}')
    print(f'  Job type: QUERY')
    print(f'  Assignee: {PROJECT_ID}')
except AlreadyExists:
    assignments = list(reservation_client.list_assignments(parent=reservation.name))
    assignment = assignments[0]
    print(f'Assignment exists: {assignment.name}')

print(f'\nReservation ready:')
print(f'  Reservation: {RESERVATION_ID} ({RESERVATION_SLOTS} slots, Enterprise)')
print(f'  Assignment:  QUERY -> {PROJECT_ID}')

# Reservation assignments take a moment to propagate
print('\nWaiting 60s for assignment to propagate...')
time.sleep(60)
print('Ready.')
```

### Create Bigtable App Profile

The [app profile](https://cloud.google.com/bigtable/docs/app-profiles) controls how requests are routed to clusters. EXPORT DATA requires:
- **Single-cluster routing** — writes go to exactly one cluster (no multi-cluster replication conflicts)
- **LOW request priority** — export traffic yields to serving reads

We use `gcloud` to create the profile since the Python client's app profile API requires more boilerplate.

```python
# Create app profile for EXPORT DATA
import subprocess

result = subprocess.run([
    'gcloud', 'bigtable', 'app-profiles', 'create', APP_PROFILE_ID,
    '--instance', BT_INSTANCE,
    '--project', PROJECT_ID,
    '--route-to', f'{BT_INSTANCE}-c1',
    '--priority', 'PRIORITY_LOW',
    '--standard',
    '--description', 'BigQuery EXPORT DATA - single cluster, LOW priority',
    '--force',
], capture_output=True, text=True)

if result.returncode == 0:
    print(f'Created app profile: {APP_PROFILE_ID}')
elif 'already exists' in result.stderr.lower():
    print(f'App profile exists: {APP_PROFILE_ID}')
else:
    print(f'Error: {result.stderr.strip()}')
    # Try without --standard flag (older gcloud versions)
    result2 = subprocess.run([
        'gcloud', 'bigtable', 'app-profiles', 'create', APP_PROFILE_ID,
        '--instance', BT_INSTANCE,
        '--project', PROJECT_ID,
        '--route-to', f'{BT_INSTANCE}-c1',
        '--description', 'BigQuery EXPORT DATA - single cluster routing',
        '--force',
    ], capture_output=True, text=True)
    if result2.returncode == 0:
        print(f'Created app profile: {APP_PROFILE_ID} (without priority flag)')
    elif 'already exists' in result2.stderr.lower():
        print(f'App profile exists: {APP_PROFILE_ID}')
    else:
        print(f'Error: {result2.stderr.strip()}')

# Verify
result = subprocess.run([
    'gcloud', 'bigtable', 'app-profiles', 'describe', APP_PROFILE_ID,
    '--instance', BT_INSTANCE,
    '--project', PROJECT_ID,
], capture_output=True, text=True)
if result.returncode == 0:
    print(f'\nProfile details:')
    for line in result.stdout.strip().split('\n'):
        print(f'  {line}')
```

## 3. First Export — Simple Types with Concatenation

The [EXPORT DATA](https://cloud.google.com/bigquery/docs/reference/standard-sql/other-statements#export_data_statement) statement [writes](https://cloud.google.com/bigtable/docs/writes) BigQuery query results directly to Bigtable. The key insight is how columns map:

- The **first column** in the SELECT becomes the Bigtable **row key**
- Each remaining column wrapped in a **STRUCT** maps to a Bigtable **[column family](https://cloud.google.com/bigtable/docs/schema-design#column-families)**, with STRUCT field names becoming **column qualifiers**

For this first export, we concatenate all features into a single pipe-delimited string stored in one cell (`features:_values`). This is the fastest pattern for ML serving because the access pattern is always "read all features for one entity" — one cell read, one decode.

### Build and Execute the Export

```python
# Bigtable URI for EXPORT DATA (includes app profile)
BIGTABLE_URI = (
    f'https://bigtable.googleapis.com/projects/{PROJECT_ID}'
    f'/instances/{BT_INSTANCE}'
    f'/appProfiles/{APP_PROFILE_ID}'
    f'/tables/{BT_TABLE}'
)

# Feature columns to export (5 INT64 + 5 FLOAT64 from dense_features)
FEATURE_COLUMNS = [
    'feature_int_1', 'feature_int_2', 'feature_int_3', 'feature_int_4', 'feature_int_5',
    'feature_float_1', 'feature_float_2', 'feature_float_3', 'feature_float_4', 'feature_float_5',
]

FEATURE_TYPES = {
    'feature_int_1': 'INT64',
    'feature_int_2': 'INT64',
    'feature_int_3': 'INT64',
    'feature_int_4': 'INT64',
    'feature_int_5': 'INT64',
    'feature_float_1': 'FLOAT64',
    'feature_float_2': 'FLOAT64',
    'feature_float_3': 'FLOAT64',
    'feature_float_4': 'FLOAT64',
    'feature_float_5': 'FLOAT64',
}

EXPORT_SQL = f"""
EXPORT DATA OPTIONS (
    format = 'CLOUD_BIGTABLE',
    uri = '{BIGTABLE_URI}',
    overwrite = true
) AS
SELECT
    CONCAT(entity_group, '#', entity_id) AS rowkey,
    STRUCT(
        CONCAT(
            CAST(feature_int_1 AS STRING), '|',
            CAST(feature_int_2 AS STRING), '|',
            CAST(feature_int_3 AS STRING), '|',
            CAST(feature_int_4 AS STRING), '|',
            CAST(feature_int_5 AS STRING), '|',
            CAST(ROUND(feature_float_1, 4) AS STRING), '|',
            CAST(ROUND(feature_float_2, 4) AS STRING), '|',
            CAST(ROUND(feature_float_3, 4) AS STRING), '|',
            CAST(ROUND(feature_float_4, 4) AS STRING), '|',
            CAST(ROUND(feature_float_5, 4) AS STRING)
        ) AS _values
    ) AS {BT_COLUMN_FAMILY}
FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
"""

print('EXPORT DATA query:')
print(EXPORT_SQL)
```

```python
# Run the EXPORT DATA query
print('Exporting features to Bigtable...')
print(f'  Source: {BQ_PROJECT}.{BQ_DATASET}.dense_features')
print(f'  Target: {BT_INSTANCE}/{BT_TABLE}')
print()

start = time.perf_counter()
job = bq.query(EXPORT_SQL)
job.result()  # Wait for completion
export_time = time.perf_counter() - start

print(f'Export complete in {export_time:.1f}s')
print(f'  Job ID:          {job.job_id}')
print(f'  Bytes processed: {job.total_bytes_processed:,}')
if job.slot_millis:
    print(f'  Slot time:       {job.slot_millis:,} ms')
```

```python
# Verify — count exported rows by scanning Bigtable
partial_rows = table.read_rows(
    filter_=CellsColumnLimitFilter(1),
    limit=10000,
)
partial_rows.consume_all()

# Count data rows (exclude any metadata rows like #schema)
data_rows = [key for key in partial_rows.rows.keys() if not key.startswith(b'#')]
print(f'Feature rows in Bigtable: {len(data_rows)}')

# Show a few sample row keys
sample_keys = sorted(data_rows)[:5]
print(f'\nSample row keys:')
for key in sample_keys:
    print(f'  {key.decode("utf-8")}')
```

## 4. Self-Describing Schema

The feature table should describe itself. We write a special `#schema` metadata row that stores column names, types, delimiter, and a version tag. The serving client reads this once at startup, builds a decoder, and uses it for all subsequent lookups. See the Bigtable [schema design](https://cloud.google.com/bigtable/docs/schema-design) guide for best practices on table structure.

This eliminates hardcoded schemas in application code. When the export query adds a feature or changes types, only the schema row needs to match — no code changes on the serving side.

### Write the Schema Metadata Row

```python
# Write the #schema metadata row
SCHEMA = {
    'columns': FEATURE_COLUMNS,
    'types': FEATURE_TYPES,
    'delimiter': '|',
    'version': 'v1',
    'source_table': f'{BQ_PROJECT}.{BQ_DATASET}.dense_features',
    'description': 'Dense features — 5 INT64 + 5 FLOAT64 features per entity',
}

schema_json = json.dumps(SCHEMA, indent=2)

row = table.direct_row('#schema')
row.set_cell(
    BT_COLUMN_FAMILY,
    '_schema',
    schema_json.encode('utf-8'),
)
row.commit()

print('Wrote #schema row to Bigtable')
print(f'  Table:     {BT_INSTANCE}/{BT_TABLE}')
print(f'  Family:    {BT_COLUMN_FAMILY}')
print(f'  Qualifier: _schema')
print(f'  Size:      {len(schema_json)} bytes')
print(f'\nSchema contents:')
for col in FEATURE_COLUMNS:
    print(f'  {col:<20s} {FEATURE_TYPES[col]}')
print(f'\nDelimiter: {SCHEMA["delimiter"]}')
print(f'Version:   {SCHEMA["version"]}')
```

```python
# Read schema back and build decoder
schema_row = table.read_row('#schema', filter_=CellsColumnLimitFilter(1))
schema_data = json.loads(
    schema_row.cells[BT_COLUMN_FAMILY][b'_schema'][0].value.decode('utf-8')
)

TYPE_PARSERS = {'INT64': int, 'FLOAT64': float, 'STRING': str}

def decode_features(raw_value, schema):
    """Decode a pipe-delimited feature string using the schema."""
    values = raw_value.split(schema['delimiter'])
    return {
        col: TYPE_PARSERS[schema['types'][col]](val)
        for col, val in zip(schema['columns'], values)
    }

print('Schema read back successfully:')
print(f'  Version:    {schema_data["version"]}')
print(f'  Columns:    {len(schema_data["columns"])}')
print(f'  Delimiter:  {schema_data["delimiter"]}')
print(f'\nDecoder ready — TYPE_PARSERS: {list(TYPE_PARSERS.keys())}')
```

---
## 5. Read Features

Now let's [read](https://cloud.google.com/bigtable/docs/reads) features back using the schema-based decoder. This is exactly what a serving pipeline would do at inference time: look up an entity's features by row key, decode using the schema, and feed them to the model.

The [Bigtable Python client](https://cloud.google.com/python/docs/reference/bigtable/latest) provides `read_row()` for single-key lookups — the fastest access pattern.

### Single Entity Read

```python
# Read our test entity's features from Bigtable
row_key = f'{TEST_ENTITY_GROUP}#{TEST_ENTITY_ID}'

bt_row = table.read_row(
    row_key.encode('utf-8'),
    filter_=CellsColumnLimitFilter(1),
)

if bt_row is None:
    print(f'No row found for {row_key}')
else:
    raw_value = bt_row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
    features = decode_features(raw_value, schema_data)

    print(f'Features for entity {row_key}:')
    print(f'  Raw value: {raw_value}')
    print()
    for col, val in features.items():
        print(f'  {col:<20s} = {val}')
```

### Batch Read — Multiple Entities

```python
# Read a batch of entities and display as a DataFrame
sample_keys = sorted(data_rows)[:20]  # First 20 entities

rows_data = []
for key in sample_keys:
    row = table.read_row(key, filter_=CellsColumnLimitFilter(1))
    if row and b'_values' in row.cells.get(BT_COLUMN_FAMILY, {}):
        raw = row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
        feats = decode_features(raw, schema_data)
        feats['entity_key'] = key.decode('utf-8')
        rows_data.append(feats)

features_df = pd.DataFrame(rows_data)
features_df = features_df[['entity_key'] + FEATURE_COLUMNS]

print(f'Sample features ({len(features_df)} entities):')
print(features_df.to_string(index=False))
```

### Other Read Methods

The [Python client library](https://cloud.google.com/python/docs/reference/bigtable/latest) is the primary API for production serving. But Bigtable supports multiple read interfaces — each suited to different use cases. Here are three more ways to read the same features.

## 6. Latency Shootout: BigQuery vs Bigtable

This is the payoff. We time the same feature lookup two ways:
1. **BigQuery** — run a SELECT query with a WHERE clause (the analytical approach)
2. **Bigtable** — [point read](https://cloud.google.com/bigtable/docs/reads) by row key + decode (the feature store approach)

We run each multiple times, alternating entities, and compare the distributions.

```python
# Benchmark: BQ query vs Bigtable read
N_ITERATIONS = 10

# Pick test entities for variety
test_keys = sorted(data_rows)[:5]

bq_latencies = []
bt_latencies = []

for i in range(N_ITERATIONS):
    entity_key = test_keys[i % len(test_keys)]
    entity_key_str = entity_key.decode('utf-8')
    parts = entity_key_str.split('#')
    e_group, e_id = parts[0], parts[1]

    # BigQuery: SELECT with WHERE clause
    bq_sql = f"""
    SELECT * FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
    WHERE entity_group = '{e_group}' AND entity_id = '{e_id}'
    """

    start = time.perf_counter()
    bq.query(bq_sql).to_dataframe()
    bq_latencies.append(time.perf_counter() - start)

    # Bigtable: point read + decode
    start = time.perf_counter()
    row = table.read_row(entity_key, filter_=CellsColumnLimitFilter(1))
    raw = row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
    feats = decode_features(raw, schema_data)
    bt_latencies.append(time.perf_counter() - start)

    print(f'  Iteration {i+1:2d}: BQ={bq_latencies[-1]:.3f}s  BT={bt_latencies[-1]*1000:.1f}ms')

bq_median_ms = sorted(bq_latencies)[len(bq_latencies)//2] * 1000
bt_median_ms = sorted(bt_latencies)[len(bt_latencies)//2] * 1000
speedup = bq_median_ms / bt_median_ms

print(f'\nMedian latency:')
print(f'  BigQuery:  {bq_median_ms:,.0f} ms')
print(f'  Bigtable:  {bt_median_ms:,.1f} ms')
print(f'  Speedup:   {speedup:,.0f}x')
```

```python
# Visualize the latency comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Panel 1: Side-by-side bar chart (log scale)
ax1 = axes[0]
medians = [bq_median_ms, bt_median_ms]
colors = ['#E53935', '#43A047']
bars = ax1.bar(
    ['BigQuery\n(SQL query)', 'Bigtable\n(point read)'],
    medians, color=colors, width=0.5, edgecolor='white',
)

for bar, val in zip(bars, medians):
    label = f'{val:,.0f} ms' if val > 100 else f'{val:.1f} ms'
    ax1.text(
        bar.get_x() + bar.get_width() / 2, bar.get_height() + max(medians) * 0.02,
        label, ha='center', va='bottom', fontweight='bold', fontsize=11,
    )

ax1.set_ylabel('Median Latency (ms)', fontweight='bold')
ax1.set_title('Feature Retrieval Latency', fontsize=13, fontweight='bold')
ax1.set_yscale('log')

# Panel 2: Distribution of Bigtable latencies (ms)
ax2 = axes[1]
bt_ms = [t * 1000 for t in bt_latencies]
ax2.hist(bt_ms, bins=max(5, N_ITERATIONS // 2), color='#43A047', alpha=0.8, edgecolor='white')
ax2.axvline(
    bt_median_ms, color='#1B5E20', linestyle='--', linewidth=2,
    label=f'Median: {bt_median_ms:.1f} ms',
)
ax2.set_xlabel('Latency (ms)', fontweight='bold')
ax2.set_ylabel('Count', fontweight='bold')
ax2.set_title('Bigtable Read Latency Distribution', fontsize=13, fontweight='bold')
ax2.legend(fontsize=10)

plt.suptitle(
    f'Feature Store Speedup: {speedup:,.0f}x Faster Than SQL',
    fontsize=14, fontweight='bold', y=1.02,
)
plt.tight_layout()
plt.show()
```

### Interpreting the Latency Gap

The difference is dramatic: BigQuery takes seconds because it must schedule a job, scan the table, and return results. Bigtable returns in milliseconds because it's a simple key-value lookup — no query planning, no aggregation, no job scheduling.

This is why feature stores exist. The model computation itself takes microseconds. Without pre-computed features, the feature retrieval dominates the total inference latency by orders of magnitude.

**The tradeoff:** BigQuery computes features *fresh* on every query. Bigtable serves *pre-computed* features that may be slightly stale. For most ML applications, features that are seconds to minutes old are perfectly fine — model quality degrades gracefully with small staleness windows.

## 7. Native Column Mapping (Quick Preview)

The concatenated-string approach works well for uniform numeric features — all features in one cell, one read, one decode. But there's an alternative: map each BigQuery column to its own Bigtable column qualifier.

Instead of CONCAT into `_values`, each STRUCT field becomes a separate column:

| Row key | `features:feature_int_1` | `features:feature_int_2` | `features:feature_float_1` | ... |
|---------|--------------------------|--------------------------|----------------------------|-----|
| `A#000001` | `42` | `7` | `3.14` | ... |

This is covered in depth in [Notebook 2: Serialization](./Bigtable%20Feature%20Store%20-%20Serialization.ipynb), which compares 5 serialization methods and their storage/latency tradeoffs. Here's a quick demo.

```python
# Quick demo — EXPORT DATA with individual column qualifiers
# This writes each feature as a separate Bigtable column qualifier
# We use a different table to avoid overwriting our concatenated export

NATIVE_TABLE = 'features-native-preview'

# Create the preview table if it doesn't exist
try:
    native_bt_table = instance.table(NATIVE_TABLE)
    native_bt_table.create(
        column_families={BT_COLUMN_FAMILY: bigtable.column_family.MaxVersionsGCRule(1)}
    )
    print(f'Created Bigtable table: {NATIVE_TABLE}')
except AlreadyExists:
    native_bt_table = instance.table(NATIVE_TABLE)
    print(f'Bigtable table exists: {NATIVE_TABLE}')

NATIVE_URI = (
    f'https://bigtable.googleapis.com/projects/{PROJECT_ID}'
    f'/instances/{BT_INSTANCE}'
    f'/appProfiles/{APP_PROFILE_ID}'
    f'/tables/{NATIVE_TABLE}'
)

NATIVE_EXPORT_SQL = f"""
EXPORT DATA OPTIONS (
    format = 'CLOUD_BIGTABLE',
    uri = '{NATIVE_URI}',
    overwrite = true
) AS
SELECT
    CONCAT(entity_group, '#', entity_id) AS rowkey,
    STRUCT(
        CAST(feature_int_1 AS STRING) AS feature_int_1,
        CAST(feature_int_2 AS STRING) AS feature_int_2,
        CAST(ROUND(feature_float_1, 4) AS STRING) AS feature_float_1,
        CAST(ROUND(feature_float_2, 4) AS STRING) AS feature_float_2
    ) AS {BT_COLUMN_FAMILY}
FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
LIMIT 100
"""

print('Running native column export (4 features, 100 rows)...')
start = time.perf_counter()
job = bq.query(NATIVE_EXPORT_SQL)
job.result()
print(f'Export complete in {time.perf_counter() - start:.1f}s')

# Read back a sample row to show the per-column structure
sample_row = native_bt_table.read_row(
    f'{TEST_ENTITY_GROUP}#{TEST_ENTITY_ID}'.encode('utf-8'),
    filter_=CellsColumnLimitFilter(1),
)

if sample_row:
    print(f'\nNative column layout for {TEST_ENTITY_GROUP}#{TEST_ENTITY_ID}:')
    for qualifier, cells in sorted(sample_row.cells.get(BT_COLUMN_FAMILY, {}).items()):
        value = cells[0].value.decode('utf-8')
        print(f'  {BT_COLUMN_FAMILY}:{qualifier.decode("utf-8"):<20s} = {value}')
else:
    print(f'No row found for {TEST_ENTITY_GROUP}#{TEST_ENTITY_ID}')
```

Each feature lives in its own column qualifier. The advantage: you can read individual features without decoding the entire row. The tradeoff: more metadata overhead per read, slightly higher latency for "read all features" patterns.

[Notebook 2: Serialization](./Bigtable%20Feature%20Store%20-%20Serialization.ipynb) measures these tradeoffs precisely.

## 9. Cleanup

The Enterprise reservation bills per-second for allocated slots. Now that the export is complete and verified, we delete the reservation stack to stop charges. The features remain in Bigtable — only the BigQuery reservation resources are removed.

> **Important:** Delete in reverse order — assignment first, then reservation. The API rejects deletion of a reservation that still has assignments.

```python
# Delete reservation resources (reverse order)
print('Cleaning up reservation resources...')

# Step 1: Delete assignment
try:
    reservation_client.delete_assignment(name=assignment.name)
    print(f'  Deleted assignment: {assignment.name}')
except NotFound:
    print(f'  Assignment already deleted')
except Exception as e:
    print(f'  Assignment cleanup: {e}')

# Step 2: Delete reservation
try:
    reservation_client.delete_reservation(name=reservation.name)
    print(f'  Deleted reservation: {reservation.name}')
except NotFound:
    print(f'  Reservation already deleted')
except Exception as e:
    print(f'  Reservation cleanup: {e}')

print('\nWaiting 30s for assignment deletion to propagate...')
time.sleep(30)
print('Reservation cleanup complete.')
```
