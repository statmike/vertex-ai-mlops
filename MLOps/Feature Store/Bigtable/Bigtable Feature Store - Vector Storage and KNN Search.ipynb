## 1. Vector Encoding Fundamentals

Bigtable stores vectors as **raw binary bytes** in a standalone column. Each float32 value occupies 4 bytes, so a 128-dimensional vector = 512 bytes. Modern embedding models produce vectors from 256 to 3,072+ dimensions (e.g., Vertex AI `text-embedding-005` = 768d, OpenAI `text-embedding-3-large` = 3,072d), meaning 1–12 KB per vector.

### Why Standalone Column?

GoogleSQL's vector distance functions (`COSINE_DISTANCE`, `EUCLIDEAN_DISTANCE`) require `TO_VECTOR32()` to convert raw bytes to `ARRAY<FLOAT32>`. This function expects the column to contain **only** the binary-encoded float32 array — no protobuf wrapping, no JSON encoding, no concatenation with other fields.

| Encoding | Works with KNN? | Why |
|----------|----------------|-----|
| Raw binary float32 (standalone column) | **Yes** | `TO_VECTOR32()` reads bytes directly |
| Protobuf-wrapped | No | Extra framing bytes confuse `TO_VECTOR32()` |
| JSON array | No | String, not binary |
| Concatenated with other fields | No | Mixed data in one column |
| Base64-encoded | No | Needs decoding first |

### Binary Encoding

Python's `struct` module handles the encoding:

```python
import struct

vector = [0.1, 0.2, 0.3, ...]  # 128 floats
binary = struct.pack(f'{len(vector)}f', *vector)  # 512 bytes

# Decode
decoded = struct.unpack(f'{VECTOR_DIM}f', binary)  # tuple of 128 floats
```

The `f` format code = float32 (4 bytes). Use little-endian by default (matches Bigtable's `TO_VECTOR32()`).

````python
# Demonstrate vector encoding and decoding
demo_vector = np.random.randn(VECTOR_DIM).astype(np.float32)
demo_vector = demo_vector / np.linalg.norm(demo_vector)  # L2 normalize

# Encode to binary
binary = struct.pack(f'{VECTOR_DIM}f', *demo_vector)

# Decode back
decoded = np.array(struct.unpack(f'{VECTOR_DIM}f', binary), dtype=np.float32)

print(f'Original vector:  {demo_vector[:5]}... ({len(demo_vector)} dims)')
print(f'Binary size:      {len(binary)} bytes ({len(binary) // 4} × 4 bytes)')
print(f'Decoded vector:   {decoded[:5]}...')
print(f'Round-trip match: {np.allclose(demo_vector, decoded)}')
print(f'L2 norm:          {np.linalg.norm(demo_vector):.6f} (normalized)')
````

## 2. Create Vector Table and Write Embeddings

We create a dedicated table with two column families:
- **`embeddings`** — stores the raw binary vector in a standalone column (`vector`)
- **`metadata`** — stores entity attributes (category, description) as regular strings

````python
# Create the vector table
vector_table = instance.table(BT_VECTOR_TABLE)

if vector_table.exists():
    print(f'Table {BT_VECTOR_TABLE} already exists — using existing table')
else:
    column_families = {
        'embeddings': None,
        'metadata': None,
    }
    vector_table.create(column_families=column_families)
    print(f'Created table: {BT_VECTOR_TABLE}')
    print(f'  Column families: embeddings, metadata')
````

````python
# Generate synthetic embeddings for NUM_ENTITIES entities
# Each entity gets a normalized 128-dim vector and a category label
np.random.seed(42)

CATEGORIES = ['electronics', 'clothing', 'food', 'books', 'sports']

# Create category centroids (so similar categories cluster together)
centroids = {cat: np.random.randn(VECTOR_DIM).astype(np.float32) for cat in CATEGORIES}
for cat in centroids:
    centroids[cat] = centroids[cat] / np.linalg.norm(centroids[cat])

write_start = time.perf_counter()
rows_to_write = []

for i in range(NUM_ENTITIES):
    entity_id = f'entity_{i:06d}'
    category = CATEGORIES[i % len(CATEGORIES)]
    
    # Generate vector: category centroid + random noise (so same-category entities are similar)
    noise = np.random.randn(VECTOR_DIM).astype(np.float32) * 0.3
    vector = centroids[category] + noise
    vector = vector / np.linalg.norm(vector)  # L2 normalize
    
    # Encode vector as raw binary float32
    vector_bytes = struct.pack(f'{VECTOR_DIM}f', *vector)
    
    row = vector_table.direct_row(entity_id.encode('utf-8'))
    row.set_cell('embeddings', 'vector', vector_bytes)
    row.set_cell('metadata', 'category', category.encode('utf-8'))
    row.set_cell('metadata', 'entity_index', str(i).encode('utf-8'))
    rows_to_write.append(row)
    
    # Batch write every 100 rows
    if len(rows_to_write) >= 100:
        vector_table.mutate_rows(rows_to_write)
        rows_to_write = []

# Write remaining rows
if rows_to_write:
    vector_table.mutate_rows(rows_to_write)

write_time = time.perf_counter() - write_start
print(f'Wrote {NUM_ENTITIES} vectors to Bigtable in {write_time:.1f}s')
print(f'  Dimensions: {VECTOR_DIM}')
print(f'  Bytes per vector: {VECTOR_DIM * 4}')
print(f'  Total vector data: {NUM_ENTITIES * VECTOR_DIM * 4 / 1024:.1f} KB')
print(f'  Categories: {CATEGORIES}')
````

````python
# Verify: read back a vector and decode it
test_row = vector_table.read_row(
    'entity_000000'.encode('utf-8'),
    filter_=CellsColumnLimitFilter(1),
)

if test_row:
    vector_bytes = test_row.cells['embeddings'][b'vector'][0].value
    decoded_vector = np.array(
        struct.unpack(f'{VECTOR_DIM}f', vector_bytes), dtype=np.float32
    )
    category = test_row.cells['metadata'][b'category'][0].value.decode('utf-8')
    
    print(f'Entity:    entity_000000')
    print(f'Category:  {category}')
    print(f'Vector:    [{", ".join(f"{v:.4f}" for v in decoded_vector[:5])}, ...] ({len(decoded_vector)} dims)')
    print(f'L2 norm:   {np.linalg.norm(decoded_vector):.6f}')
    print(f'Bytes:     {len(vector_bytes)}')
else:
    print('ERROR: Could not read back test entity')
````

## 3. KNN Search with GoogleSQL

[GoogleSQL for Bigtable](https://cloud.google.com/bigtable/docs/googlesql-overview) supports vector distance functions that enable KNN search directly in SQL:

- **`COSINE_DISTANCE(v1, v2)`** — measures angular similarity (0 = identical, 2 = opposite)
- **`EUCLIDEAN_DISTANCE(v1, v2)`** — measures absolute distance in vector space

Both use `TO_VECTOR32()` to convert the raw binary bytes to `ARRAY<FLOAT32>` at query time.

### How KNN Works in Bigtable

This is **brute-force KNN**: the query scans every row in the table, computes the distance to the target vector, sorts, and returns the top K. There is no pre-built index — query time is O(n) where n = number of rows.

```sql
SELECT _key,
       COSINE_DISTANCE(
           TO_VECTOR32(embeddings['vector']),
           <target_vector>
       ) AS distance
FROM `features-vectors`
ORDER BY distance
LIMIT 10
```

````python
# Read a query entity's vector to use as the search target
query_entity = 'entity_000000'
query_row = vector_table.read_row(
    query_entity.encode('utf-8'),
    filter_=CellsColumnLimitFilter(1),
)

query_vector_bytes = query_row.cells['embeddings'][b'vector'][0].value
query_vector = np.array(
    struct.unpack(f'{VECTOR_DIM}f', query_vector_bytes), dtype=np.float32
)
query_category = query_row.cells['metadata'][b'category'][0].value.decode('utf-8')

print(f'Query entity: {query_entity} (category: {query_category})')
print(f'Query vector: [{", ".join(f"{v:.4f}" for v in query_vector[:5])}, ...]')

# Build the GoogleSQL query
# TO_VECTOR32 expects the binary bytes; we pass the target as a float array literal
vector_literal = ', '.join(f'{v:.6f}' for v in query_vector)

COSINE_SQL = f"""
SELECT
    _key,
    COSINE_DISTANCE(
        TO_VECTOR32(embeddings['vector']),
        ARRAY<FLOAT32>[{vector_literal}]
    ) AS distance
FROM `{BT_VECTOR_TABLE}`
ORDER BY distance
LIMIT 10
"""

print(f'\nGoogleSQL KNN query (first 200 chars):')
print(COSINE_SQL[:200] + '...')
print(f'\nRun this query in Bigtable Studio:')
print(f'  https://console.cloud.google.com/bigtable/instances/{BT_INSTANCE}/studio?project={PROJECT_ID}')
````

````python
# Python-based KNN search (for environments where GoogleSQL is not available)
# This reads all vectors and computes distances client-side

def knn_search(table, query_vector, k=10, distance='cosine'):
    """Brute-force KNN search by reading all rows and computing distances."""
    results = []
    
    read_start = time.perf_counter()
    rows = list(table.read_rows(filter_=CellsColumnLimitFilter(1)))
    read_time = time.perf_counter() - read_start
    
    compute_start = time.perf_counter()
    for row in rows:
        if b'vector' not in row.cells.get('embeddings', {}):
            continue
        
        vector_bytes = row.cells['embeddings'][b'vector'][0].value
        vector = np.array(
            struct.unpack(f'{VECTOR_DIM}f', vector_bytes), dtype=np.float32
        )
        
        if distance == 'cosine':
            dist = 1 - np.dot(query_vector, vector) / (
                np.linalg.norm(query_vector) * np.linalg.norm(vector)
            )
        else:  # euclidean
            dist = np.linalg.norm(query_vector - vector)
        
        category = row.cells.get('metadata', {}).get(b'category', [None])
        cat_val = category[0].value.decode('utf-8') if category[0] else 'unknown'
        
        results.append({
            'key': row.row_key.decode('utf-8'),
            'distance': float(dist),
            'category': cat_val,
        })
    
    compute_time = time.perf_counter() - compute_start
    
    results.sort(key=lambda x: x['distance'])
    return results[:k], read_time, compute_time

# Run KNN search
print(f'Searching for {10} nearest neighbors of {query_entity} (category: {query_category})...')
print()

neighbors, read_time, compute_time = knn_search(vector_table, query_vector, k=10)

print(f'{"Rank":<6} {"Entity":<18} {"Category":<14} {"Cosine Distance":<16}')
print('-' * 54)
for i, n in enumerate(neighbors):
    marker = ' ← query' if n['key'] == query_entity else ''
    print(f'{i+1:<6} {n["key"]:<18} {n["category"]:<14} {n["distance"]:<16.6f}{marker}')

print(f'\nLatency:')
print(f'  Read {NUM_ENTITIES} vectors: {read_time*1000:.0f}ms')
print(f'  Compute distances:  {compute_time*1000:.0f}ms')
print(f'  Total:              {(read_time+compute_time)*1000:.0f}ms')

# Check: do most neighbors share the query's category?
same_cat = sum(1 for n in neighbors if n['category'] == query_category)
print(f'\n{same_cat}/10 neighbors share category "{query_category}" (expected: most, since vectors cluster by category)')
````

````python
# Compare cosine vs euclidean distance
cosine_neighbors, _, _ = knn_search(vector_table, query_vector, k=10, distance='cosine')
euclid_neighbors, _, _ = knn_search(vector_table, query_vector, k=10, distance='euclidean')

print(f'{"Rank":<6} {"Cosine KNN":<18} {"Cosine Dist":<14} {"Euclidean KNN":<18} {"Euclidean Dist":<14}')
print('-' * 70)
for i in range(10):
    c = cosine_neighbors[i]
    e = euclid_neighbors[i]
    print(f'{i+1:<6} {c["key"]:<18} {c["distance"]:<14.6f} {e["key"]:<18} {e["distance"]:<14.6f}')

# Check overlap
cosine_keys = {n['key'] for n in cosine_neighbors}
euclid_keys = {n['key'] for n in euclid_neighbors}
overlap = len(cosine_keys & euclid_keys)
print(f'\nOverlap: {overlap}/10 entities appear in both top-10 lists')
print('For normalized vectors, cosine and euclidean rankings are equivalent.')
````

## 4. Feature Store Use Cases for Vector Search

### Similar Entity Lookup

Find users/items most similar to a given entity — useful for:
- **Collaborative filtering**: "users who are similar to this user also liked..."
- **Content-based recommendations**: "products with similar embeddings"
- **Fraud detection**: "transactions whose embedding is far from all known-good clusters"

### Candidate Retrieval

KNN search as the first stage of a two-stage retrieval system:
1. **Retrieve**: KNN finds the top-100 candidate items from Bigtable
2. **Rank**: A more expensive model (neural network, gradient boosted trees) re-ranks the candidates

### Limitations

| Aspect | Bigtable KNN | Dedicated Vector DB (e.g., Vertex AI Matching Engine) |
|--------|-------------|------------------------------------------------------|
| **Algorithm** | Brute-force KNN (scans all rows) | ANN with pre-built index (e.g., ScaNN, HNSW) |
| **Query time** | O(n) — linear in table size | O(log n) — sublinear |
| **Suitable scale** | Up to ~100K vectors | Millions to billions |
| **Index maintenance** | None required | Index rebuild on updates |
| **Consistency** | Real-time (reads latest data) | Index may lag behind writes |
| **Additional infrastructure** | None (uses existing Bigtable) | Separate service |

**When Bigtable KNN is enough:**
- Feature store with < 100K entities where you need occasional similarity searches
- Low-QPS similarity queries (not high-throughput retrieval)
- You already have Bigtable and don't want another service

**When to use a dedicated vector DB:**
- Millions+ vectors requiring sub-millisecond retrieval
- High-throughput retrieval (thousands of queries per second)
- ANN with tunable recall/latency tradeoffs

## 5. Performance Characteristics

KNN search performance depends on table size and vector dimensionality. Let's measure both.

````python
# Benchmark: KNN latency vs number of rows scanned
# We'll search across subsets of increasing size

subset_sizes = [100, 250, 500, 750, NUM_ENTITIES]
latency_results = []

for size in subset_sizes:
    # Read only the first `size` rows
    rows = list(vector_table.read_rows(
        filter_=CellsColumnLimitFilter(1),
        limit=size,
    ))
    
    # Compute distances
    start = time.perf_counter()
    distances = []
    for row in rows:
        if b'vector' not in row.cells.get('embeddings', {}):
            continue
        vec_bytes = row.cells['embeddings'][b'vector'][0].value
        vec = np.array(struct.unpack(f'{VECTOR_DIM}f', vec_bytes), dtype=np.float32)
        dist = 1 - np.dot(query_vector, vec) / (np.linalg.norm(query_vector) * np.linalg.norm(vec))
        distances.append(dist)
    
    compute_ms = (time.perf_counter() - start) * 1000
    latency_results.append({'rows': size, 'compute_ms': compute_ms})
    print(f'  {size:>5} rows: {compute_ms:>8.1f}ms compute')

results_df = pd.DataFrame(latency_results)

fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(results_df['rows'], results_df['compute_ms'], 'o-', color='#4285F4', linewidth=2)
ax.set_xlabel('Number of Rows Scanned')
ax.set_ylabel('Compute Time (ms)')
ax.set_title(f'KNN Compute Time vs Table Size ({VECTOR_DIM}-dim vectors)')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f'\nKNN scales linearly with table size — O(n) brute-force scan.')
print(f'For {NUM_ENTITIES} rows at {VECTOR_DIM} dims, compute is ~{results_df.iloc[-1]["compute_ms"]:.0f}ms.')
print(f'Network read time adds to this — total latency is read + compute.')
````

````python
# Benchmark: impact of vector dimensionality
# Generate vectors at different dimensions and measure distance computation

dim_results = []
test_sizes = [32, 128, 256, 768, 1536, 3072]
n_vectors = 1000

for dim in test_sizes:
    vectors = np.random.randn(n_vectors, dim).astype(np.float32)
    vectors = vectors / np.linalg.norm(vectors, axis=1, keepdims=True)
    q = vectors[0]
    
    start = time.perf_counter()
    for v in vectors:
        _ = 1 - np.dot(q, v) / (np.linalg.norm(q) * np.linalg.norm(v))
    compute_ms = (time.perf_counter() - start) * 1000
    
    dim_results.append({
        'dimensions': dim,
        'compute_ms': compute_ms,
        'bytes_per_vector': dim * 4,
    })

dim_df = pd.DataFrame(dim_results)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(range(len(test_sizes)), dim_df['compute_ms'], color='#34A853', alpha=0.8)
axes[0].set_xticks(range(len(test_sizes)))
axes[0].set_xticklabels(test_sizes)
axes[0].set_xlabel('Vector Dimensions')
axes[0].set_ylabel('Compute Time (ms)')
axes[0].set_title(f'Distance Computation Time ({n_vectors} vectors)')

axes[1].bar(range(len(test_sizes)), dim_df['bytes_per_vector'], color='#FBBC04', alpha=0.8)
axes[1].set_xticks(range(len(test_sizes)))
axes[1].set_xticklabels(test_sizes)
axes[1].set_xlabel('Vector Dimensions')
axes[1].set_ylabel('Bytes per Vector')
axes[1].set_title('Storage Cost per Vector')

plt.tight_layout()
plt.show()

print(f'\nHigher dimensions increase both compute time and storage:')
for _, r in dim_df.iterrows():
    print(f'  {int(r["dimensions"]):>4}d: {r["compute_ms"]:>6.1f}ms compute, {int(r["bytes_per_vector"]):>5} bytes/vector')
````

## 6. Combining Vectors with Regular Features

In a real feature store, entities have both embedding vectors (for similarity search) and scalar features (for prediction). The key insight is that vectors go in a **standalone column** in the `embeddings` family, while scalar features use the regular `features` family — same table, same row key, single read to get both.

```
Row key: entity_000001
├── embeddings:vector  →  [raw binary float32 × 128]    ← standalone, for KNN
├── features:_values   →  "29|4136|91|...|0.3935"       ← concatenated scalars
└── metadata:category  →  "electronics"                 ← string metadata
```

This pattern means a serving endpoint can:
1. Read a single row to get both the embedding and scalar features
2. Use the embedding for similarity-based retrieval
3. Use the scalar features for model prediction
4. All in one Bigtable read (~5ms)

````python
# Demonstrate reading vectors and metadata in a single row read
sample_row = vector_table.read_row(
    'entity_000042'.encode('utf-8'),
    filter_=CellsColumnLimitFilter(1),
)

if sample_row:
    # Vector
    vec_bytes = sample_row.cells['embeddings'][b'vector'][0].value
    vec = np.array(struct.unpack(f'{VECTOR_DIM}f', vec_bytes), dtype=np.float32)
    
    # Metadata
    category = sample_row.cells['metadata'][b'category'][0].value.decode('utf-8')
    idx = sample_row.cells['metadata'][b'entity_index'][0].value.decode('utf-8')
    
    print(f'Entity:    entity_000042')
    print(f'Category:  {category}')
    print(f'Index:     {idx}')
    print(f'Vector:    [{", ".join(f"{v:.4f}" for v in vec[:5])}, ...] ({len(vec)} dims)')
    print(f'\nSingle read returned both vector ({len(vec_bytes)} bytes) and metadata.')
    print(f'A production table would also include a features column family with scalar features.')
````

## Summary

This notebook demonstrated vector storage and KNN search in Bigtable:

| Topic | Key Takeaway |
|-------|-------------|
| **Vector encoding** | Raw binary float32 in a standalone column — `struct.pack(f'{n}f', *vector)` |
| **Storage** | 4 bytes per dimension (128-dim = 512 B, 768-dim = 3 KB, 3072-dim = 12 KB per vector) |
| **KNN search** | `COSINE_DISTANCE(TO_VECTOR32(...), target)` via GoogleSQL |
| **Algorithm** | Brute-force scan — O(n), no index required |
| **Scale** | Suitable for up to ~100K vectors; beyond that, use a dedicated vector DB |
| **Combined storage** | Vectors + scalar features in one row, one read |

### When to Use Bigtable for Vectors

Bigtable KNN works well when:
- You already have a Bigtable feature store and want to add similarity search
- Your entity count is moderate (< 100K)
- You need real-time consistency (no index lag)
- Query volume is low to moderate

For high-throughput similarity search over millions of vectors, consider [Vertex AI Vector Search](https://cloud.google.com/vertex-ai/docs/vector-search/overview) or other dedicated ANN services.
