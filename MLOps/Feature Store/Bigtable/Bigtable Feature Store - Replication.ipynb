# Bigtable Feature Store — Replication

Multi-cluster [replication](https://cloud.google.com/bigtable/docs/replication-overview) provides high availability and disaster recovery for your feature store. This notebook creates a separate **Production** instance with two clusters in different zones, demonstrates replication behavior, and cleans up afterward.

> **Why a separate instance?** The main `feature-store` instance from Notebook 0 is a Development instance (single cluster, 1 node). Development instances [cannot have replication](https://cloud.google.com/bigtable/docs/instances-clusters-nodes#development-instances). This notebook creates and manages its own Production instance so we can demonstrate replication without affecting the shared resources.

**Prerequisites:** Run [Notebook 0: Environment](./Bigtable%20Feature%20Store%20-%20Environment.ipynb) first (for API enablement and authentication).

````python
PROJECT_ID = 'statmike-mlops-349915'
REGION = 'us-central1'
SERIES = 'bigtable-feature-store'

# Replication instance (Production, multi-cluster)
BT_REP_INSTANCE = 'feature-store-replicated'
BT_REP_TABLE = 'features-rep'
BT_COLUMN_FAMILY = 'features'

# Cluster locations (two different zones for zone-level HA)
CLUSTER_1_ID = f'{BT_REP_INSTANCE}-c1'
CLUSTER_1_ZONE = 'us-central1-a'
CLUSTER_1_NODES = 1

CLUSTER_2_ID = f'{BT_REP_INSTANCE}-c2'
CLUSTER_2_ZONE = 'us-central1-b'
CLUSTER_2_NODES = 1

print(f'Project:    {PROJECT_ID}')
print(f'Instance:   {BT_REP_INSTANCE} (Production, 2 clusters)')
print(f'Cluster 1:  {CLUSTER_1_ID} ({CLUSTER_1_ZONE}, {CLUSTER_1_NODES} node)')
print(f'Cluster 2:  {CLUSTER_2_ID} ({CLUSTER_2_ZONE}, {CLUSTER_2_NODES} node)')
````

## 1. Why Replication for Feature Stores

A feature store is on the critical path of every prediction request. If Bigtable becomes unavailable, predictions fail. [Replication](https://cloud.google.com/bigtable/docs/replication-overview) adds redundancy:

| Benefit | How replication helps |
|---------|---------------------|
| **Zone-level HA** | If one zone goes down, the other cluster continues serving |
| **Regional DR** | Cross-region replication survives regional outages |
| **Lower latency** | Multi-cluster routing sends reads to the nearest cluster |
| **Read scaling** | Multiple clusters share the read load |

### The Consistency Trade-off

| Routing mode | Consistency | Failover | Use case |
|-------------|-------------|----------|----------|
| **Single-cluster** | Strong (within cluster) | Manual (update app profile) | Financial features, fraud detection — correctness over availability |
| **Multi-cluster** | Eventual | Automatic | Recommendation features, user profiles — availability over strict consistency |

For most feature stores, **eventual consistency is fine**: if a user's features are a few seconds behind, the prediction quality barely changes. But for features where staleness means wrong answers (e.g., account balance for fraud detection), use single-cluster routing with a manual failover runbook.

## 2. Create Production Instance with Replication

Development instances have a single cluster. To add replication, we need a [Production instance](https://cloud.google.com/bigtable/docs/instances-clusters-nodes#production-instances) with at least 1 node per cluster.

> **Cost note:** Production instances are billed per node-hour. We create a minimal setup (1 node per cluster = 2 nodes total) and delete the instance at the end of this notebook.

````python
# Create Production instance with two clusters
rep_instance = bt_client.instance(
    BT_REP_INSTANCE,
    display_name='Feature Store Replicated',
    instance_type=enums.Instance.Type.PRODUCTION,
)

if rep_instance.exists():
    rep_instance.reload()
    clusters, _ = rep_instance.list_clusters()
    print(f'Instance {BT_REP_INSTANCE} already exists ({len(clusters)} clusters)')
else:
    cluster_1 = rep_instance.cluster(
        CLUSTER_1_ID,
        location_id=CLUSTER_1_ZONE,
        serve_nodes=CLUSTER_1_NODES,
    )
    
    operation = rep_instance.create(clusters=[cluster_1])
    operation.result()
    print(f'Created Production instance: {BT_REP_INSTANCE}')
    print(f'  Cluster 1: {CLUSTER_1_ID} ({CLUSTER_1_ZONE}, {CLUSTER_1_NODES} node)')
    
    # Add second cluster for replication
    cluster_2 = rep_instance.cluster(
        CLUSTER_2_ID,
        location_id=CLUSTER_2_ZONE,
        serve_nodes=CLUSTER_2_NODES,
    )
    operation = cluster_2.create()
    operation.result()
    print(f'  Cluster 2: {CLUSTER_2_ID} ({CLUSTER_2_ZONE}, {CLUSTER_2_NODES} node)')

# Verify
rep_instance.reload()
clusters, _ = rep_instance.list_clusters()
print(f'\nInstance: {BT_REP_INSTANCE}')
instance_type = {0: 'UNSPECIFIED', 1: 'PRODUCTION', 2: 'DEVELOPMENT'}.get(rep_instance.type_, 'UNKNOWN')
print(f'  Type: {instance_type}')
for c in clusters:
    print(f'  Cluster: {c.cluster_id} ({c.location_id}, {c.serve_nodes} nodes)')
````

````python
# Create a table with a features column family
rep_table = rep_instance.table(BT_REP_TABLE)

if rep_table.exists():
    print(f'Table {BT_REP_TABLE} already exists')
else:
    rep_table.create(column_families={BT_COLUMN_FAMILY: None, 'metadata': None})
    print(f'Created table: {BT_REP_TABLE}')
    print(f'  Column families: {BT_COLUMN_FAMILY}, metadata')
````

````python
# Write sample data (100 entities)
import random
random.seed(42)

rows_to_write = []
for i in range(100):
    entity_key = f'A#{str(i).zfill(6)}'
    values = '|'.join([
        str(random.randint(1, 1000)),
        str(random.randint(1, 10000)),
        str(random.randint(1, 100)),
        str(round(random.uniform(0, 100), 4)),
        str(round(random.uniform(0, 1), 4)),
    ])
    
    row = rep_table.direct_row(entity_key.encode('utf-8'))
    row.set_cell(BT_COLUMN_FAMILY, '_values', values.encode('utf-8'))
    row.set_cell('metadata', 'source', b'replication_demo')
    rows_to_write.append(row)

rep_table.mutate_rows(rows_to_write)
print(f'Wrote {len(rows_to_write)} entities to {BT_REP_TABLE}')
````

## 3. App Profiles for Replication

[App profiles](https://cloud.google.com/bigtable/docs/app-profiles) control how reads and writes are routed to clusters. With replication, you have two routing options:

| Profile type | Routing | Consistency | Failover |
|-------------|---------|-------------|----------|
| **Single-cluster** | All traffic to one cluster | Strong within that cluster | Manual — update the profile |
| **Multi-cluster** | Automatic routing to nearest healthy cluster | Eventual | Automatic — transparent to client |

````python
# Create app profiles for different routing strategies

# Profile 1: Single-cluster routing (strong consistency)
result = subprocess.run([
    'gcloud', 'bigtable', 'app-profiles', 'create', 'single-cluster',
    '--instance', BT_REP_INSTANCE,
    '--project', PROJECT_ID,
    '--route-to', CLUSTER_1_ID,
    '--description', 'Single-cluster routing for strong consistency',
    '--force',
], capture_output=True, text=True)

if result.returncode == 0:
    print(f'Created app profile: single-cluster (routes to {CLUSTER_1_ID})')
elif 'already exists' in result.stderr.lower():
    print(f'App profile exists: single-cluster')
else:
    print(f'Error: {result.stderr.strip()[:200]}')

# Profile 2: Multi-cluster routing (automatic failover)
result = subprocess.run([
    'gcloud', 'bigtable', 'app-profiles', 'create', 'multi-cluster',
    '--instance', BT_REP_INSTANCE,
    '--project', PROJECT_ID,
    '--route-any',
    '--description', 'Multi-cluster routing for automatic failover',
    '--force',
], capture_output=True, text=True)

if result.returncode == 0:
    print(f'Created app profile: multi-cluster (routes to any cluster)')
elif 'already exists' in result.stderr.lower():
    print(f'App profile exists: multi-cluster')
else:
    print(f'Error: {result.stderr.strip()[:200]}')

# List all profiles
result = subprocess.run([
    'gcloud', 'bigtable', 'app-profiles', 'list',
    '--instance', BT_REP_INSTANCE,
    '--project', PROJECT_ID,
    '--format', 'table(name.basename(), singleClusterRouting.clusterId, multiClusterRoutingUseAny)',
], capture_output=True, text=True)
print(f'\nApp profiles for {BT_REP_INSTANCE}:')
print(result.stdout)
````

---
## 4. Replication in Action

Data written to one cluster is automatically replicated to the other. Let's measure the replication lag by writing to cluster 1 and reading from cluster 2.

````python
# Read via each cluster profile to verify data is replicated
# Create table references with specific app profiles
table_c1 = bigtable.Client(project=PROJECT_ID, admin=False).instance(
    BT_REP_INSTANCE
).table(BT_REP_TABLE, app_profile_id='single-cluster')

table_mc = bigtable.Client(project=PROJECT_ID, admin=False).instance(
    BT_REP_INSTANCE
).table(BT_REP_TABLE, app_profile_id='multi-cluster')

# Read the same entity through each profile
test_key = 'A#000001'

print(f'Reading entity {test_key} through different app profiles:\n')

for label, tbl in [('single-cluster', table_c1), ('multi-cluster', table_mc)]:
    t0 = time.perf_counter()
    row = tbl.read_row(test_key.encode('utf-8'), filter_=CellsColumnLimitFilter(1))
    latency = (time.perf_counter() - t0) * 1000
    
    if row:
        values = row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
        print(f'  {label:>16}: {values} ({latency:.1f}ms)')
    else:
        print(f'  {label:>16}: NOT FOUND ({latency:.1f}ms)')
````

````python
# Measure replication lag: write a new row, then poll for it on the other cluster
print('Measuring replication lag...\n')

repl_key = 'REPL_TEST#lag_measure'
repl_value = f'written_at_{time.time():.0f}'

# Write via the default profile (goes to cluster 1 via single-cluster routing)
row = rep_table.direct_row(repl_key.encode('utf-8'))
row.set_cell(BT_COLUMN_FAMILY, '_values', repl_value.encode('utf-8'))
row.commit()
write_time = time.perf_counter()
print(f'Wrote: {repl_key} = {repl_value}')

# Poll via multi-cluster routing (may hit cluster 2)
attempts = 0
replicated = False
while time.perf_counter() - write_time < 30:
    attempts += 1
    check_row = table_mc.read_row(
        repl_key.encode('utf-8'),
        filter_=CellsColumnLimitFilter(1),
    )
    if check_row:
        lag = (time.perf_counter() - write_time) * 1000
        read_value = check_row.cells[BT_COLUMN_FAMILY][b'_values'][0].value.decode('utf-8')
        print(f'Replicated in {lag:.0f}ms (attempt {attempts})')
        print(f'  Read value: {read_value}')
        replicated = True
        break
    time.sleep(0.1)

if not replicated:
    print(f'Replication took longer than 30s (or routing hit the same cluster)')

# Cleanup test row
row = rep_table.direct_row(repl_key.encode('utf-8'))
row.delete()
row.commit()
print(f'\nDeleted test row: {repl_key}')
print('\nNote: Replication is typically sub-second for same-region clusters.')
print('Cross-region replication has higher latency (100ms-1s+).')
````

---
## 5. Failover Patterns

### Multi-Cluster Routing (Automatic Failover)

With multi-cluster routing, Bigtable automatically routes requests to the nearest healthy cluster. If one cluster becomes unavailable, traffic seamlessly shifts to the other — no application changes needed.

### Single-Cluster Routing (Manual Failover)

With single-cluster routing, all traffic goes to one cluster. To fail over, update the app profile to point to a different cluster:

```bash
# Failover: switch from cluster 1 to cluster 2
gcloud bigtable app-profiles update single-cluster \
    --instance=feature-store-replicated \
    --project=PROJECT_ID \
    --route-to=feature-store-replicated-c2 \
    --force
```

### Choosing a Failover Strategy

| Strategy | Pros | Cons | Best for |
|----------|------|------|----------|
| **Multi-cluster** | Zero-touch failover, lower latency | Eventual consistency, conflict resolution | Most feature stores |
| **Single-cluster** | Strong consistency | Manual failover, brief downtime | Financial/compliance features |

For feature stores, **multi-cluster routing is the default recommendation**. The eventual consistency window (typically < 1 second within a region) is negligible for ML serving, and automatic failover eliminates an operational burden.

---
## 6. Replication Topology Patterns

| Topology | Clusters | Use Case | Consistency |
|----------|----------|----------|-------------|
| **Single zone** | 1 cluster | Development, non-critical workloads | N/A (no replication) |
| **Multi-zone (same region)** | 2+ clusters in different zones | Zone-level HA, standard production | Eventual (sub-second within region) |
| **Multi-region** | 2+ clusters in different regions | Regional DR, global serving | Eventual (100ms-1s+ cross-region) |

### Multi-Zone (This Demo)

```
us-central1-a          us-central1-b
┌──────────────┐      ┌──────────────┐
│  Cluster 1   │◄────►│  Cluster 2   │
│  (1 node)    │ repl │  (1 node)    │
└──────────────┘      └──────────────┘
       ▲                     ▲
       │    App Profile      │
       │  (multi-cluster)    │
       └─────────┬───────────┘
                 │
          Client Request
```

### Multi-Region (Reference)

For global serving (e.g., features served from US, EU, and APAC), add clusters in multiple regions:

```bash
# Add a cluster in europe-west1
gcloud bigtable clusters create feature-store-rep-eu \
    --instance=feature-store-replicated \
    --project=PROJECT_ID \
    --zone=europe-west1-b \
    --num-nodes=1
```

Cross-region replication increases data durability but adds replication lag proportional to the geographic distance.

## Cleanup

Delete the replicated Production instance. This removes all clusters, tables, and data. The main `feature-store` Development instance from NB0 is not affected.

> **Important:** Production instances incur per-node-hour charges. Always delete when done with this demo.

````python
# Delete app profiles first (optional — deleting the instance removes them)
for profile in ['single-cluster', 'multi-cluster']:
    result = subprocess.run([
        'gcloud', 'bigtable', 'app-profiles', 'delete', profile,
        '--instance', BT_REP_INSTANCE,
        '--project', PROJECT_ID,
        '--force', '--quiet',
    ], capture_output=True, text=True)
    if result.returncode == 0:
        print(f'Deleted app profile: {profile}')
    else:
        print(f'App profile {profile}: {result.stderr.strip()[:100]}')

# Delete the instance (removes all clusters and tables)
rep_instance = bt_client.instance(BT_REP_INSTANCE)
if rep_instance.exists():
    rep_instance.delete()
    print(f'\nDeleted instance: {BT_REP_INSTANCE}')
    print('  All clusters, tables, and data have been removed.')
    print('  No ongoing charges.')
else:
    print(f'Instance {BT_REP_INSTANCE} does not exist — nothing to clean up')
````

## Summary

This notebook demonstrated multi-cluster replication for a Bigtable feature store:

| Topic | Key Takeaway |
|-------|-------------|
| **Instance type** | Replication requires a Production instance (Development instances are single-cluster) |
| **Cluster topology** | Two zones in the same region for zone-level HA; cross-region for DR |
| **Multi-cluster routing** | Automatic failover, eventual consistency — recommended for most feature stores |
| **Single-cluster routing** | Strong consistency, manual failover — for compliance-sensitive features |
| **Replication lag** | Sub-second within a region; 100ms-1s+ cross-region |
| **Cost** | Per-node-hour per cluster — 2 clusters = 2× the node cost |

### When to Replicate

| Scenario | Replication? | Why |
|----------|-------------|-----|
| Development / prototyping | No | Use Development instance (cheaper, simpler) |
| Production with SLA | **Yes** | Zone-level HA prevents outages from impacting predictions |
| Global serving | **Yes** | Multi-region reduces read latency for geographically distributed clients |
| Compliance (data residency) | **Maybe** | Cross-region replication may conflict with data residency requirements |
