# Bigtable Feature Store — Environment

This notebook creates the shared environment for the Bigtable Feature Store series. It generates synthetic data in [BigQuery](https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types) with full data type coverage and creates a [Cloud Bigtable](https://cloud.google.com/bigtable/docs/overview) instance for online serving.

**What this notebook creates:**
- BigQuery dataset with two tables:
  - **Dense table** (~130,000 rows) — 1 row per entity, 200+ features covering every BQ data type
  - **Sparse table** (~2.6M rows) — multiple rows per entity with timestamps for time-series features
- Cloud Bigtable development instance with a feature table

**Compound entity key:** All data uses a two-part key (`entity_group` + `entity_id`) designed for the Bigtable [row key](https://cloud.google.com/bigtable/docs/schema-design) `{entity_group}#{entity_id}`.

## Dense Table Generation

The dense table provides **one row per entity** with 200+ feature columns covering every [BigQuery data type](https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types). This is the primary table the other notebooks use for feature store sync operations.

### Design

| Aspect | Value |
|--------|-------|
| **Entity groups** | 26 groups ("A" through "Z") |
| **Entities per group** | 5,000 (IDs: "000001" through "005000") |
| **Total rows** | 130,000 |
| **Compound key** | `entity_group` + `entity_id` (Bigtable row key: `{entity_group}#{entity_id}`) |

### Data Type Coverage

| BQ Data Type | Column Prefix | Count |
|-------------|---------------|-------|
| INT64 | `feature_int_` | 5 |
| FLOAT64 | `feature_float_` | 180 |
| NUMERIC | `feature_numeric_` | 3 |
| BIGNUMERIC | `feature_bignumeric_` | 2 |
| STRING | `feature_string_` | 5 |
| BYTES | `feature_bytes_` | 2 |
| BOOL | `feature_bool_` | 3 |
| DATE | `feature_date_` | 3 |
| DATETIME | `feature_datetime_` | 2 |
| TIME | `feature_time_` | 2 |
| TIMESTAMP | `feature_timestamp_` | 3 |
| GEOGRAPHY | `feature_geo_` | 2 |
| JSON | `feature_json_` | 2 |
| ARRAY<INT64> | `feature_array_int_` | 2 |
| ARRAY<STRING> | `feature_array_string_` | 1 |
| STRUCT | `feature_struct_` | 2 |
| ARRAY<STRUCT> | `feature_array_struct_` | 1 |
| RANGE<DATE> | `feature_range_date_` | 1 |

The 180 `feature_float_` columns use varied distributions (uniform, scaled, offset) to provide realistic bulk feature data. Combined with the typed columns above, the table exceeds 200 total columns.

````python
# Build and execute the dense table SQL
# The ~175 extra FLOAT64 columns are generated programmatically

# Float columns (feature_float_6 through feature_float_180)
float_columns = []
for i in range(6, 181):
    if i % 5 == 0:
        float_columns.append(f'    ROUND(RAND() * 1000, 4) AS feature_float_{i}')
    elif i % 5 == 1:
        float_columns.append(f'    ROUND(RAND() * 100 - 50, 4) AS feature_float_{i}')
    elif i % 5 == 2:
        float_columns.append(f'    ROUND(RAND() * 10, 4) AS feature_float_{i}')
    elif i % 5 == 3:
        float_columns.append(f'    ROUND(RAND() * 500 + 100, 4) AS feature_float_{i}')
    else:
        float_columns.append(f'    ROUND(RAND(), 6) AS feature_float_{i}')

float_cols_sql = ',\n'.join(float_columns)

dense_query = f"""
CREATE OR REPLACE TABLE `{BQ_PROJECT}.{BQ_DATASET}.dense_features` AS
WITH
entity_groups AS (
    SELECT CHR(code) AS entity_group
    FROM UNNEST(GENERATE_ARRAY(65, 90)) AS code
),
entity_ids AS (
    SELECT FORMAT('%06d', id) AS entity_id
    FROM UNNEST(GENERATE_ARRAY(1, 5000)) AS id
),
entities AS (
    SELECT entity_group, entity_id
    FROM entity_groups CROSS JOIN entity_ids
)
SELECT
    entity_group,
    entity_id,

    -- INT64 features (5)
    CAST(FLOOR(RAND() * 1000) AS INT64) AS feature_int_1,
    CAST(FLOOR(RAND() * 10000) AS INT64) AS feature_int_2,
    CAST(FLOOR(RAND() * 100) AS INT64) AS feature_int_3,
    CAST(FLOOR(RAND() * 1000000) AS INT64) AS feature_int_4,
    CAST(MOD(ABS(FARM_FINGERPRINT(CONCAT(entity_group, entity_id))), 500) AS INT64) AS feature_int_5,

    -- FLOAT64 features (first 5 — the rest are generated below)
    ROUND(RAND() * 100, 4) AS feature_float_1,
    ROUND(RAND() * 1000, 4) AS feature_float_2,
    ROUND(RAND() * 50 - 25, 4) AS feature_float_3,
    ROUND(RAND() * 10, 4) AS feature_float_4,
    ROUND(RAND(), 6) AS feature_float_5,

    -- NUMERIC features (3)
    CAST(ROUND(RAND() * 99999, 2) AS NUMERIC) AS feature_numeric_1,
    CAST(ROUND(RAND() * 9999999, 4) AS NUMERIC) AS feature_numeric_2,
    CAST(ROUND(RAND() * 100, 6) AS NUMERIC) AS feature_numeric_3,

    -- BIGNUMERIC features (2)
    CAST(ROUND(RAND() * 999999999999, 2) AS BIGNUMERIC) AS feature_bignumeric_1,
    CAST(ROUND(RAND() * 99999999, 8) AS BIGNUMERIC) AS feature_bignumeric_2,

    -- STRING features (5)
    CONCAT('cat_', CAST(CAST(FLOOR(RAND() * 20) AS INT64) AS STRING)) AS feature_string_1,
    CONCAT('type_', CAST(CAST(FLOOR(RAND() * 50) AS INT64) AS STRING)) AS feature_string_2,
    ['alpha', 'beta', 'gamma', 'delta', 'epsilon', 'zeta', 'eta', 'theta'][ORDINAL(CAST(CEIL(RAND() * 8) AS INT64))] AS feature_string_3,
    CONCAT(entity_group, '_', ['north', 'south', 'east', 'west'][ORDINAL(CAST(CEIL(RAND() * 4) AS INT64))]) AS feature_string_4,
    FORMAT('%s-%06d', entity_group, CAST(FLOOR(RAND() * 999999) AS INT64)) AS feature_string_5,

    -- BYTES features (2)
    CAST(CONCAT('bytes_', entity_group, entity_id) AS BYTES) AS feature_bytes_1,
    CAST(FORMAT('%06d', CAST(FLOOR(RAND() * 999999) AS INT64)) AS BYTES) AS feature_bytes_2,

    -- BOOL features (3)
    RAND() > 0.5 AS feature_bool_1,
    RAND() > 0.7 AS feature_bool_2,
    MOD(CAST(FLOOR(RAND() * 10) AS INT64), 2) = 0 AS feature_bool_3,

    -- DATE features (3)
    DATE_ADD(DATE '2020-01-01', INTERVAL CAST(FLOOR(RAND() * 1825) AS INT64) DAY) AS feature_date_1,
    DATE_ADD(DATE '2022-06-01', INTERVAL CAST(FLOOR(RAND() * 730) AS INT64) DAY) AS feature_date_2,
    DATE_ADD(DATE '2023-01-01', INTERVAL CAST(FLOOR(RAND() * 365) AS INT64) DAY) AS feature_date_3,

    -- DATETIME features (2)
    DATETIME_ADD(DATETIME '2023-01-01 00:00:00', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_datetime_1,
    DATETIME_ADD(DATETIME '2024-01-01 00:00:00', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_datetime_2,

    -- TIME features (2)
    TIME_ADD(TIME '00:00:00', INTERVAL CAST(FLOOR(RAND() * 86400) AS INT64) SECOND) AS feature_time_1,
    TIME_ADD(TIME '00:00:00', INTERVAL CAST(FLOOR(RAND() * 86400) AS INT64) SECOND) AS feature_time_2,

    -- TIMESTAMP features (3)
    TIMESTAMP_ADD(TIMESTAMP '2023-01-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_timestamp_1,
    TIMESTAMP_ADD(TIMESTAMP '2024-01-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_timestamp_2,
    TIMESTAMP_ADD(TIMESTAMP '2024-06-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 262800) AS INT64) MINUTE) AS feature_timestamp_3,

    -- GEOGRAPHY features (2)
    ST_GEOGPOINT(
        ROUND(-180 + RAND() * 360, 6),
        ROUND(-90 + RAND() * 180, 6)
    ) AS feature_geo_1,
    ST_GEOGPOINT(
        ROUND(-122.5 + RAND() * 5, 6),
        ROUND(37.0 + RAND() * 3, 6)
    ) AS feature_geo_2,

    -- JSON features (2)
    JSON_OBJECT(
        'score', ROUND(RAND() * 100, 2),
        'label', CONCAT('class_', CAST(CAST(FLOOR(RAND() * 10) AS INT64) AS STRING)),
        'active', RAND() > 0.5
    ) AS feature_json_1,
    JSON_OBJECT(
        'x', ROUND(RAND() * 10, 4),
        'y', ROUND(RAND() * 10, 4),
        'tag', entity_group
    ) AS feature_json_2,

    -- ARRAY<INT64> features (2)
    [CAST(FLOOR(RAND() * 100) AS INT64), CAST(FLOOR(RAND() * 100) AS INT64), CAST(FLOOR(RAND() * 100) AS INT64)] AS feature_array_int_1,
    [CAST(FLOOR(RAND() * 1000) AS INT64), CAST(FLOOR(RAND() * 1000) AS INT64)] AS feature_array_int_2,

    -- ARRAY<STRING> features (1)
    [
        ['red', 'blue', 'green', 'yellow'][ORDINAL(CAST(CEIL(RAND() * 4) AS INT64))],
        ['circle', 'square', 'triangle'][ORDINAL(CAST(CEIL(RAND() * 3) AS INT64))]
    ] AS feature_array_string_1,

    -- STRUCT features (2)
    STRUCT(
        ROUND(RAND() * 100, 2) AS value,
        CONCAT('unit_', CAST(CAST(FLOOR(RAND() * 5) AS INT64) AS STRING)) AS unit,
        RAND() > 0.5 AS is_valid
    ) AS feature_struct_1,
    STRUCT(
        CAST(FLOOR(RAND() * 1000) AS INT64) AS count,
        ROUND(RAND(), 4) AS ratio
    ) AS feature_struct_2,

    -- ARRAY<STRUCT> features (1)
    [
        STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name),
        STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name),
        STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name)
    ] AS feature_array_struct_1,

    -- RANGE<DATE> features (1)
    RANGE(
        DATE_ADD(DATE '2020-01-01', INTERVAL CAST(FLOOR(RAND() * 1000) AS INT64) DAY),
        DATE_ADD(DATE '2023-01-01', INTERVAL CAST(FLOOR(RAND() * 730) AS INT64) DAY)
    ) AS feature_range_date_1,

    -- Bulk FLOAT64 features (feature_float_6 through feature_float_180)
{float_cols_sql}

FROM entities
"""

print(f'Dense table SQL: {len(dense_query):,} characters')
print('Executing CREATE OR REPLACE TABLE ...')

bq.query(dense_query).result()

print('Dense table created successfully.')
````

## Sparse Table Generation

The sparse table provides **multiple rows per entity** with timestamps, simulating time-series event data. Each entity has ~20 events spread over the past 90 days.

### Design

| Aspect | Value |
|--------|-------|
| **Entities** | Same 130,000 from dense table |
| **Events per entity** | 20 |
| **Total rows** | 2,600,000 |
| **Time range** | 90 days of history |
| **Event types** | purchase, view, click, search, return |
| **Metrics** | 5 continuous FLOAT64 columns |

This table is used by later notebooks to compute time-windowed aggregations (e.g., "count of purchases in last 7 days") and to test streaming feature updates.

````python
# Generate the sparse events table
print('Creating sparse_events table (~2.6M rows)...')
print('This may take 30-60 seconds.\n')

sparse_query = f"""
CREATE OR REPLACE TABLE `{BQ_PROJECT}.{BQ_DATASET}.sparse_events` AS
WITH
entities AS (
    SELECT entity_group, entity_id
    FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
),
events_per_entity AS (
    SELECT
        entity_group,
        entity_id,
        event_num
    FROM entities
    CROSS JOIN UNNEST(GENERATE_ARRAY(1, 20)) AS event_num
)
SELECT
    entity_group,
    entity_id,
    TIMESTAMP_SUB(
        CURRENT_TIMESTAMP(),
        INTERVAL CAST(FLOOR(RAND() * 90 * 24 * 60) AS INT64) MINUTE
    ) AS event_timestamp,
    ['purchase', 'view', 'click', 'search', 'return'][ORDINAL(CAST(CEIL(RAND() * 5) AS INT64))] AS event_type,
    ROUND(RAND() * 100, 2) AS metric_1,
    ROUND(RAND() * 50, 2) AS metric_2,
    ROUND(RAND() * 200, 2) AS metric_3,
    ROUND(RAND() * 10, 2) AS metric_4,
    ROUND(RAND() * 1000, 2) AS metric_5,
    IF(RAND() > 0.9, 'positive', 'negative') AS label
FROM events_per_entity
"""

bq.query(sparse_query).result()
print('Sparse events table created.')
````

### Verify Sparse Table

## Bigtable Instance & Table

### Cloud Bigtable Architecture

[Cloud Bigtable](https://cloud.google.com/bigtable/docs/overview) organizes resources in a hierarchy that maps directly to how data is stored, served, and protected:

```
Project
└── Instance          ← logical container (billing, access control)
    ├── Cluster 1     ← physical compute + storage in a specific zone
    │   └── Nodes     ← serve read/write requests
    ├── Cluster 2     ← optional: another zone for replication
    │   └── Nodes
    └── ...
        └── Tables    ← live inside the instance, replicated across clusters
```

| Concept | What It Is | Documentation |
|---------|-----------|---------------|
| **[Instance](https://cloud.google.com/bigtable/docs/instances-clusters-nodes)** | Logical container for clusters and tables. All access control, monitoring, and billing happen at this level. | [Instances, clusters, nodes](https://cloud.google.com/bigtable/docs/instances-clusters-nodes) |
| **Cluster** | A group of nodes in a single Google Cloud zone. Each cluster stores a full copy of the instance's data. | [Creating clusters](https://cloud.google.com/bigtable/docs/instances-clusters-nodes#clusters) |
| **Nodes** | Compute resources that handle read/write traffic. More nodes = higher throughput. Bigtable distributes tablets (row ranges) across nodes automatically. | [Node scaling](https://cloud.google.com/bigtable/docs/instances-clusters-nodes#nodes) |
| **[Replication](https://cloud.google.com/bigtable/docs/replication-overview)** | Multi-cluster instances replicate data automatically. Provides higher availability and read throughput. Eventually consistent between clusters (typically seconds). | [Replication overview](https://cloud.google.com/bigtable/docs/replication-overview) |
| **[App Profile](https://cloud.google.com/bigtable/docs/app-profiles)** | Routing policy that controls which cluster handles a request. Required for multi-cluster instances. Also sets priority levels (HIGH/MEDIUM/LOW) to isolate workloads. | [App profiles](https://cloud.google.com/bigtable/docs/app-profiles) |

**Instance types:**
- **Production** — 3+ nodes, SLA-backed, auto-scales. For real workloads.
- **[Development](https://cloud.google.com/bigtable/docs/instances-clusters-nodes#development)** — 1 node, no SLA, no replication. For learning and prototyping. Can be upgraded to production later.

We use a **development instance** for this series to keep costs minimal while demonstrating all the same APIs and patterns. Everything you learn here — table design, column families, reads, writes, exports — works identically on production instances.

### App Profiles — Why They Matter

Every Bigtable API request is routed through an [app profile](https://cloud.google.com/bigtable/docs/app-profiles). The default profile works for single-cluster instances, but app profiles become essential for:

1. **Traffic isolation** — Assign LOW priority to batch exports (`EXPORT DATA`) so they don't starve online serving reads that need HIGH priority.
2. **Cluster routing** — In multi-cluster setups, route reads to the nearest cluster and writes to a specific cluster for consistency.
3. **Workload identification** — Monitor which app profiles generate which traffic patterns in Cloud Monitoring dashboards.

We create dedicated app profiles in **Notebook 1 (Fundamentals)** for export operations and explore production app profile patterns in **Notebook 7 (Schema Evolution & Operations)**.
