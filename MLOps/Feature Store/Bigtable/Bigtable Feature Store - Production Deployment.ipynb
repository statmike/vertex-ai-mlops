# Bigtable Feature Store — Production Deployment

Enterprise feature stores need two things beyond the core functionality: **reproducible infrastructure** (Terraform) and **polyglot client support** (Go, Java, Node.js — not just Python).

This notebook demonstrates both:

1. **Terraform** provisions a complete Bigtable feature store — instance, table, column families, garbage collection policies, and app profiles — using declarative configuration. We run the full `init → plan → apply → modify → destroy` lifecycle.

2. **Multi-language clients** read features from the Terraform-provisioned instance using Go, Java, and Node.js, alongside the Python baseline used throughout this series.

The result is a production-ready deployment pattern: infrastructure as code, with serving clients in whatever language your team uses.

**Prerequisites:** Run [Notebook 0: Environment](./Bigtable%20Feature%20Store%20-%20Environment.ipynb) first (for GCP project setup and API enablement). This notebook creates its own Bigtable instance — it does not depend on the `feature-store` instance from NB0.

> **Cross-reference:** [Notebook 7 (Schema Evolution & Operations)](./Bigtable%20Feature%20Store%20-%20Schema%20Evolution%20and%20Operations.ipynb), Section 11 shows the same Terraform resource blocks as a reference pattern. This notebook makes them executable. [Notebook 8 (Serving Integration)](./Bigtable%20Feature%20Store%20-%20Serving%20Integration.ipynb), Section 3 compares five read methods including REST/gRPC. This notebook adds executable Go, Java, and Node.js examples.

```python
import os
import subprocess as _sp

def _get_project():
    for var in ['GOOGLE_CLOUD_PROJECT', 'PROJECT_ID', 'GCLOUD_PROJECT']:
        val = os.environ.get(var)
        if val:
            return val
    result = _sp.run(['gcloud', 'config', 'get-value', 'project'], capture_output=True, text=True)
    if result.returncode == 0 and result.stdout.strip():
        return result.stdout.strip()
    raise RuntimeError('Set PROJECT_ID or run: gcloud config set project <your-project>')

PROJECT_ID = _get_project()
REGION = 'us-central1'
ZONE = f'{REGION}-a'

BT_INSTANCE_TF = 'feature-store-tf'
BT_TABLE_TF = 'features'
BT_COLUMN_FAMILY = 'features'
BT_METADATA_FAMILY = 'metadata'

TF_DIR = 'terraform-feature-store'

print(f'Project:     {PROJECT_ID}')
print(f'Region:      {REGION}')
print(f'Zone:        {ZONE}')
print(f'Instance:    {BT_INSTANCE_TF}')
print(f'Table:       {BT_TABLE_TF}')
print(f'TF dir:      {TF_DIR}')
```

## 1. Install Terraform

[Terraform](https://www.terraform.io/) is a declarative infrastructure-as-code tool. You write *what* you want (a Bigtable instance with SSD storage, autoscaling 3–10 nodes, two column families) and Terraform figures out *how* to create it. Changes are version-controlled, code-reviewed, and applied through a plan/apply workflow that shows exactly what will change before anything happens.

For feature stores, this matters because:

- **Reproducibility**: spin up identical dev/staging/prod environments from the same config
- **Drift detection**: `terraform plan` tells you if someone changed infrastructure outside of Terraform
- **Change tracking**: git history shows who changed what, when, and why
- **Rollback**: revert a `.tf` file and `terraform apply` to undo a change

We'll download the Terraform binary, write a configuration, and provision a complete Bigtable feature store.

```python
import platform
import zipfile
import urllib.request
import stat

TERRAFORM_VERSION = '1.15.2'

arch = 'amd64' if platform.machine() == 'x86_64' else 'arm64'
tf_url = f'https://releases.hashicorp.com/terraform/{TERRAFORM_VERSION}/terraform_{TERRAFORM_VERSION}_linux_{arch}.zip'
tf_bin = pathlib.Path('terraform')

if tf_bin.exists():
    print(f'Terraform already exists at {tf_bin.resolve()}')
else:
    print(f'Downloading Terraform {TERRAFORM_VERSION} for linux_{arch}...')
    zip_path, _ = urllib.request.urlretrieve(tf_url)
    with zipfile.ZipFile(zip_path, 'r') as z:
        z.extract('terraform', '.')
    tf_bin.chmod(tf_bin.stat().st_mode | stat.S_IEXEC)
    pathlib.Path(zip_path).unlink()
    print(f'Installed to {tf_bin.resolve()}')

result = subprocess.run([str(tf_bin.resolve()), 'version'], capture_output=True, text=True)
print(result.stdout.strip())

TF = str(tf_bin.resolve())
```

---
## 2. Terraform Configuration

A production Terraform setup splits configuration across multiple files:

| File | Purpose |
|------|---------|
| `variables.tf` | Input parameters (project, region, instance name) |
| `main.tf` | Resource definitions (instance, table, GC policies, app profiles) |
| `outputs.tf` | Values to display after `apply` (instance name, project) |

We'll write all three, then walk through the `init → plan → apply` workflow.

```python
tf_dir = pathlib.Path(TF_DIR)
tf_dir.mkdir(exist_ok=True)

variables_tf = f'''variable "project_id" {{
  description = "GCP project ID"
  type        = string
  default     = "{PROJECT_ID}"
}}

variable "zone" {{
  description = "Bigtable cluster zone"
  type        = string
  default     = "{ZONE}"
}}

variable "instance_name" {{
  description = "Bigtable instance name"
  type        = string
  default     = "{BT_INSTANCE_TF}"
}}

variable "table_name" {{
  description = "Bigtable table name"
  type        = string
  default     = "{BT_TABLE_TF}"
}}
'''

(tf_dir / 'variables.tf').write_text(variables_tf)
print('variables.tf:')
print(variables_tf)
```

```python
main_tf = '''terraform {
  required_providers {
    google = {
      source  = "hashicorp/google"
      version = "~> 6.0"
    }
  }
}

provider "google" {
  project = var.project_id
}

# --- Bigtable Instance ---
resource "google_bigtable_instance" "feature_store" {
  name                = var.instance_name
  deletion_protection = false

  cluster {
    cluster_id   = "${var.instance_name}-c1"
    zone         = var.zone
    storage_type = "SSD"
    num_nodes    = 1
  }
}

# --- Feature Table ---
resource "google_bigtable_table" "features" {
  name          = var.table_name
  instance_name = google_bigtable_instance.feature_store.name

  column_family { family = "features" }
  column_family { family = "metadata" }
}

# --- Garbage Collection: keep only latest version ---
resource "google_bigtable_gc_policy" "features_gc" {
  instance_name = google_bigtable_instance.feature_store.name
  table         = google_bigtable_table.features.name
  column_family = "features"

  max_version {
    number = 1
  }
}

resource "google_bigtable_gc_policy" "metadata_gc" {
  instance_name = google_bigtable_instance.feature_store.name
  table         = google_bigtable_table.features.name
  column_family = "metadata"

  max_version {
    number = 1
  }
}

# --- App Profile for serving traffic ---
resource "google_bigtable_app_profile" "serving" {
  instance       = google_bigtable_instance.feature_store.name
  app_profile_id = "serving"
  description    = "Low-latency serving reads"

  single_cluster_routing {
    cluster_id                 = "${var.instance_name}-c1"
    allow_transactional_writes = false
  }

  ignore_warnings = true
}
'''

(tf_dir / 'main.tf').write_text(main_tf)
print('main.tf:')
print(main_tf)
```

```python
outputs_tf = '''output "instance_name" {
  value       = google_bigtable_instance.feature_store.name
  description = "Bigtable instance name"
}

output "table_name" {
  value       = google_bigtable_table.features.name
  description = "Bigtable table name"
}

output "serving_app_profile" {
  value       = google_bigtable_app_profile.serving.app_profile_id
  description = "App profile for serving reads"
}

output "project_id" {
  value       = var.project_id
  description = "GCP project ID"
}
'''

(tf_dir / 'outputs.tf').write_text(outputs_tf)
print('outputs.tf:')
print(outputs_tf)
```

## 3. Terraform Workflow

The Terraform lifecycle follows three steps:

1. **`terraform init`** — downloads the Google Cloud provider plugin
2. **`terraform plan`** — shows what will be created, modified, or destroyed (dry run)
3. **`terraform apply`** — executes the plan and creates real infrastructure

This is the same workflow whether you're provisioning a single Development instance or a multi-region Production deployment. The configuration changes; the workflow doesn't.

```python
result = subprocess.run(
    [TF, 'init'],
    cwd=TF_DIR,
    capture_output=True, text=True,
)
print(result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr)
```

```python
result = subprocess.run(
    [TF, 'plan', '-no-color'],
    cwd=TF_DIR,
    capture_output=True, text=True,
)
print(result.stdout[-3000:] if len(result.stdout) > 3000 else result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr[-2000:])
```

The plan output shows exactly what Terraform will create:

- **`google_bigtable_instance.feature_store`** — a Development instance with SSD storage
- **`google_bigtable_table.features`** — a table with `features` and `metadata` column families
- **`google_bigtable_gc_policy`** — max 1 version per cell (latest value only)
- **`google_bigtable_app_profile.serving`** — single-cluster routing for low-latency reads

Nothing has been created yet. Terraform only acts on `apply`.

```python
print('Applying Terraform configuration...')
print('This creates a Bigtable instance — may take 1-2 minutes.\n')

result = subprocess.run(
    [TF, 'apply', '-auto-approve', '-no-color'],
    cwd=TF_DIR,
    capture_output=True, text=True,
    timeout=300,
)
print(result.stdout[-3000:] if len(result.stdout) > 3000 else result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr[-2000:])
```

```python
client = bigtable.Client(project=PROJECT_ID, admin=True)
instance = client.instance(BT_INSTANCE_TF)

table = instance.table(BT_TABLE_TF)
families = table.list_column_families()

print(f'Instance:  {instance.instance_id}')
print(f'Table:     {table.table_id}')
print(f'Families:  {sorted(families.keys())}')

for name, cf in sorted(families.items()):
    gc_rule = cf.gc_rule
    print(f'  {name}: GC = {gc_rule}')

print(f'\nTerraform state matches live infrastructure.')
```

## 4. Managing Infrastructure Changes

Terraform tracks the current state of your infrastructure. When you modify the `.tf` files, `terraform plan` shows the exact diff — what will be added, changed, or destroyed. This is the production workflow for evolving a feature store: propose a change, review the plan, apply.

Let's add a `streaming` column family for real-time feature updates (the pattern from [Notebook 5: Streaming and Direct Writes](./Bigtable%20Feature%20Store%20-%20Streaming%20and%20Direct%20Writes.ipynb)).

```python
main_tf_path = pathlib.Path(TF_DIR) / 'main.tf'
current = main_tf_path.read_text()

addition = '''
# --- Streaming column family (added for real-time features) ---
resource "google_bigtable_gc_policy" "streaming_gc" {
  instance_name = google_bigtable_instance.feature_store.name
  table         = google_bigtable_table.features.name
  column_family = "streaming"

  max_version {
    number = 3
  }
}
'''

# Add streaming family to the table resource
modified = current.replace(
    '  column_family { family = "metadata" }',
    '  column_family { family = "metadata" }\n  column_family { family = "streaming" }'
)

modified += addition
main_tf_path.write_text(modified)
print('Modified main.tf — added streaming column family with max 3 versions')
```

```python
result = subprocess.run(
    [TF, 'plan', '-no-color'],
    cwd=TF_DIR,
    capture_output=True, text=True,
)
print(result.stdout[-3000:] if len(result.stdout) > 3000 else result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr[-2000:])
```

```python
result = subprocess.run(
    [TF, 'apply', '-auto-approve', '-no-color'],
    cwd=TF_DIR,
    capture_output=True, text=True,
    timeout=120,
)
print(result.stdout[-2000:] if len(result.stdout) > 2000 else result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr[-2000:])

# Verify
families = instance.table(BT_TABLE_TF).list_column_families()
print(f'\nColumn families after change: {sorted(families.keys())}')
```

## 5. Write Sample Data

Before testing multi-language clients, we need data in the Terraform-provisioned instance. We'll write 100 entities with the same compound key format (`entity_group#entity_id`) and feature schema used throughout this series.

```python
import random
random.seed(42)

table = instance.table(BT_TABLE_TF)
rows = []

# Schema row
schema_row = table.direct_row('#schema')
schema = {
    'columns': ['feature_int_1', 'feature_int_2', 'feature_int_3',
                'feature_float_1', 'feature_float_2', 'feature_float_3'],
    'types': ['INT64', 'INT64', 'INT64', 'FLOAT64', 'FLOAT64', 'FLOAT64'],
    'version': '1',
}
schema_row.set_cell('metadata', '_schema', json.dumps(schema))
schema_row.set_cell('metadata', '_version', 'v1')
rows.append(schema_row)

# 100 entities: 10 groups (A-J) x 10 entities each
groups = [chr(ord('A') + i) for i in range(10)]
for group in groups:
    for eid in range(1, 11):
        row_key = f'{group}#{eid:06d}'
        row = table.direct_row(row_key)
        
        values = [
            str(random.randint(0, 1000)),
            str(random.randint(0, 500)),
            str(random.randint(0, 100)),
            str(round(random.uniform(0, 1), 4)),
            str(round(random.uniform(-10, 10), 4)),
            str(round(random.uniform(0, 100), 4)),
        ]
        combined = '|'.join(values)
        row.set_cell('features', '_values', combined)
        
        for col_name, val in zip(schema['columns'], values):
            row.set_cell('features', col_name, val)
        
        row.set_cell('metadata', '_last_updated', datetime.datetime.now(datetime.timezone.utc).isoformat())
        rows.append(row)

# Batch write
BATCH_SIZE = 50
for i in range(0, len(rows), BATCH_SIZE):
    batch = rows[i:i + BATCH_SIZE]
    status = table.mutate_rows(batch)
    failed = sum(1 for s in status if s.code != 0)
    if failed:
        print(f'Batch {i // BATCH_SIZE}: {failed} failures')

print(f'Wrote {len(rows)} rows ({len(rows) - 1} entities + 1 schema row)')

# Verify
test_row = table.read_row('A#000001')
values = test_row.cells['features'][b'_values'][0].value.decode()
print(f'Verify A#000001: {values}')
```

## 6. Multi-Language Clients

Bigtable provides client libraries for [Python](https://cloud.google.com/python/docs/reference/bigtable/latest), [Go](https://pkg.go.dev/cloud.google.com/go/bigtable), [Java](https://cloud.google.com/java/docs/reference/google-cloud-bigtable/latest/overview), [Node.js](https://cloud.google.com/nodejs/docs/reference/bigtable/latest), [C++](https://cloud.google.com/cpp/docs/reference/bigtable/latest), and [C#](https://cloud.google.com/dotnet/docs/reference/Google.Cloud.Bigtable.V2/latest). All wrap the same gRPC API — the read patterns are structurally identical across languages.

We'll read the same entity (`A#000001`) from each language and compare. The Python read is the baseline — we've used it throughout this series. Go, Java, and Node.js are the three most common serving languages for feature stores.

> **Cross-reference:** [Notebook 8 (Serving Integration)](./Bigtable%20Feature%20Store%20-%20Serving%20Integration.ipynb), Section 3 compares five read methods (Python client, cbt CLI, GoogleSQL, BQ external table, REST/gRPC). This notebook adds executable code for Go, Java, and Node.js.

```python
print('=== Python (baseline) ===\n')

start = time.time()
row = table.read_row(
    'A#000001',
    filter_=row_filters.CellsColumnLimitFilter(1),
)
elapsed_ms = (time.time() - start) * 1000

print(f'Row key: A#000001')
print(f'Read latency: {elapsed_ms:.1f} ms\n')

for family, columns in sorted(row.cells.items()):
    fam = family.decode() if isinstance(family, bytes) else family
    print(f'Family: {fam}')
    for col, versions in sorted(columns.items()):
        c = col.decode() if isinstance(col, bytes) else col
        val = versions[0].value
        val = val.decode() if isinstance(val, bytes) else str(val)
        if len(val) > 60:
            val = val[:60] + '...'
        print(f'  {c} = {val}')
    print()
```

### Go

The [Go client](https://pkg.go.dev/cloud.google.com/go/bigtable) (`cloud.google.com/go/bigtable`) is gRPC-native and commonly used for low-latency serving endpoints. Go's compiled binaries and goroutine model make it a natural fit for high-throughput feature serving.

```python
go_dir = pathlib.Path('go-bigtable-client')
go_dir.mkdir(exist_ok=True)

go_program = f'''package main

import (
\t"context"
\t"fmt"
\t"strings"
\t"time"

\t"cloud.google.com/go/bigtable"
)

func main() {{
\tctx := context.Background()
\tprojectID := "{PROJECT_ID}"
\tinstanceID := "{BT_INSTANCE_TF}"
\ttableID := "{BT_TABLE_TF}"

\tclient, err := bigtable.NewClient(ctx, projectID, instanceID)
\tif err != nil {{
\t\tfmt.Printf("Error creating client: %v\\n", err)
\t\treturn
\t}}
\tdefer client.Close()

\ttbl := client.Open(tableID)

\tstart := time.Now()
\trow, err := tbl.ReadRow(ctx, "A#000001",
\t\tbigtable.RowFilter(bigtable.LatestNFilter(1)))
\telapsed := time.Since(start)

\tif err != nil {{
\t\tfmt.Printf("Error reading row: %v\\n", err)
\t\treturn
\t}}

\tfmt.Println("=== Go client ===")
\tfmt.Println()
\tfmt.Printf("Row key: A#000001\\n")
\tfmt.Printf("Read latency: %.1f ms\\n\\n", float64(elapsed.Microseconds())/1000.0)

\tfor family, items := range row {{
\t\tfmt.Printf("Family: %s\\n", family)
\t\tfor _, item := range items {{
\t\t\tcol := item.Column
\t\t\tparts := strings.SplitN(col, ":", 2)
\t\t\tif len(parts) == 2 {{
\t\t\t\tcol = parts[1]
\t\t\t}}
\t\t\tval := string(item.Value)
\t\t\tif len(val) > 60 {{
\t\t\t\tval = val[:60] + "..."
\t\t\t}}
\t\t\tfmt.Printf("  %s = %s\\n", col, val)
\t\t}}
\t\tfmt.Println()
\t}}
}}
'''

(go_dir / 'main.go').write_text(go_program)

# Initialize Go module
result = subprocess.run(
    ['go', 'mod', 'init', 'bigtable-client'],
    cwd=str(go_dir), capture_output=True, text=True,
)

print('Go program written. Installing dependencies...')
result = subprocess.run(
    ['go', 'get', 'cloud.google.com/go/bigtable'],
    cwd=str(go_dir), capture_output=True, text=True,
    timeout=120,
)
if result.returncode != 0:
    print(f'go get error: {result.stderr[:500]}')
else:
    print('Dependencies installed.')

result = subprocess.run(
    ['go', 'mod', 'tidy'],
    cwd=str(go_dir), capture_output=True, text=True,
    timeout=60,
)
print('Go module ready.')
```

```python
print('Running Go client...\n')

result = subprocess.run(
    ['go', 'run', 'main.go'],
    cwd=str(go_dir),
    capture_output=True, text=True,
    timeout=60,
)
print(result.stdout)
if result.returncode != 0:
    print(f'Error: {result.stderr[:1000]}')
```

### Java

The [Java client](https://cloud.google.com/java/docs/reference/google-cloud-bigtable/latest/overview) (`com.google.cloud:google-cloud-bigtable`) is the most mature Bigtable client library. It powers many large-scale production feature stores, especially in organizations with JVM-based serving infrastructure.

We use [Maven](https://maven.apache.org/) to resolve dependencies and compile the program.

```python
java_dir = pathlib.Path('java-bigtable-client')
java_dir.mkdir(exist_ok=True)
(java_dir / 'src' / 'main' / 'java').mkdir(parents=True, exist_ok=True)

pom_xml = f'''<?xml version="1.0" encoding="UTF-8"?>
<project xmlns="http://maven.apache.org/POM/4.0.0"
         xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance"
         xsi:schemaLocation="http://maven.apache.org/POM/4.0.0 http://maven.apache.org/xsd/maven-4.0.0.xsd">
    <modelVersion>4.0.0</modelVersion>
    <groupId>com.example</groupId>
    <artifactId>bigtable-client</artifactId>
    <version>1.0</version>
    <packaging>jar</packaging>

    <dependencies>
        <dependency>
            <groupId>com.google.cloud</groupId>
            <artifactId>google-cloud-bigtable</artifactId>
            <version>2.56.0</version>
        </dependency>
    </dependencies>

    <build>
        <plugins>
            <plugin>
                <groupId>org.codehaus.mojo</groupId>
                <artifactId>exec-maven-plugin</artifactId>
                <version>3.5.0</version>
                <configuration>
                    <mainClass>ReadFeatures</mainClass>
                </configuration>
            </plugin>
        </plugins>
    </build>
</project>
'''

java_program = f'''import com.google.cloud.bigtable.data.v2.BigtableDataClient;
import com.google.cloud.bigtable.data.v2.BigtableDataSettings;
import com.google.cloud.bigtable.data.v2.models.Row;
import com.google.cloud.bigtable.data.v2.models.RowCell;
import com.google.cloud.bigtable.data.v2.models.Filters;

public class ReadFeatures {{{{
    public static void main(String[] args) throws Exception {{{{
        String projectId = "{PROJECT_ID}";
        String instanceId = "{BT_INSTANCE_TF}";
        String tableId = "{BT_TABLE_TF}";

        BigtableDataSettings settings = BigtableDataSettings.newBuilder()
            .setProjectId(projectId)
            .setInstanceId(instanceId)
            .build();

        try (BigtableDataClient client = BigtableDataClient.create(settings)) {{{{
            long start = System.nanoTime();
            Row row = client.readRow(tableId, "A#000001",
                Filters.FILTERS.limit().cellsPerColumn(1));
            double elapsedMs = (System.nanoTime() - start) / 1_000_000.0;

            System.out.println("=== Java client ===");
            System.out.println();
            System.out.println("Row key: A#000001");
            System.out.printf("Read latency: %.1f ms%n%n", elapsedMs);

            String currentFamily = "";
            for (RowCell cell : row.getCells()) {{{{
                String family = cell.getFamily();
                if (!family.equals(currentFamily)) {{{{
                    if (!currentFamily.isEmpty()) System.out.println();
                    System.out.println("Family: " + family);
                    currentFamily = family;
                }}}}
                String val = cell.getValue().toStringUtf8();
                if (val.length() > 60) val = val.substring(0, 60) + "...";
                System.out.println("  " + cell.getQualifier().toStringUtf8() + " = " + val);
            }}}}
            System.out.println();
        }}}}
    }}}}
}}}}
'''

(java_dir / 'pom.xml').write_text(pom_xml)
(java_dir / 'src' / 'main' / 'java' / 'ReadFeatures.java').write_text(java_program)

# Check if Maven is available
mvn = shutil.which('mvn')
MAVEN_AVAILABLE = mvn is not None

if not MAVEN_AVAILABLE:
    print('Maven not found. Downloading Maven...')
    MVN_VERSION = '3.9.15'
    mvn_url = f'https://dlcdn.apache.org/maven/maven-3/{MVN_VERSION}/binaries/apache-maven-{MVN_VERSION}-bin.tar.gz'
    subprocess.run(['curl', '-sL', mvn_url, '-o', 'maven.tar.gz'], check=True)
    subprocess.run(['tar', 'xzf', 'maven.tar.gz'], check=True)
    mvn = str(pathlib.Path(f'apache-maven-{MVN_VERSION}/bin/mvn').resolve())
    pathlib.Path('maven.tar.gz').unlink()
    MAVEN_AVAILABLE = True
    print(f'Maven {MVN_VERSION} installed.')
else:
    print(f'Maven found: {mvn}')

if MAVEN_AVAILABLE:
    print('\nCompiling Java program (downloading dependencies — may take 1-2 minutes)...')
    result = subprocess.run(
        [mvn, 'compile', '-q', '-B'],
        cwd=str(java_dir),
        capture_output=True, text=True,
        timeout=300,
    )
    if result.returncode != 0:
        print(f'Compilation error: {result.stderr[:1000]}')
        MAVEN_AVAILABLE = False
    else:
        print('Java program compiled successfully.')
```

```python
if MAVEN_AVAILABLE:
    print('Running Java client...\n')
    result = subprocess.run(
        [mvn, 'exec:java', '-q', '-B',
         '-Dexec.mainClass=ReadFeatures'],
        cwd=str(java_dir),
        capture_output=True, text=True,
        timeout=60,
    )
    print(result.stdout)
    if result.returncode != 0:
        print(f'Error: {result.stderr[:1000]}')
else:
    print('=== Java client (reference only) ===')
    print()
    print('Maven was not available to compile the program.')
    print('The Java source code is at: java-bigtable-client/src/main/java/ReadFeatures.java')
    print()
    print('To run it manually:')
    print('  cd java-bigtable-client')
    print('  mvn compile exec:java -Dexec.mainClass=ReadFeatures')
```

### Node.js

The [Node.js client](https://cloud.google.com/nodejs/docs/reference/bigtable/latest) (`@google-cloud/bigtable`) uses async/await and is popular for serverless feature serving (Cloud Functions, Cloud Run with Node.js). The non-blocking I/O model handles concurrent feature reads efficiently.

```python
node_dir = pathlib.Path('node-bigtable-client')
node_dir.mkdir(exist_ok=True)

node_program = f'''const {{Bigtable}} = require("@google-cloud/bigtable");

async function main() {{
  const bigtable = new Bigtable({{projectId: "{PROJECT_ID}"}});
  const instance = bigtable.instance("{BT_INSTANCE_TF}");
  const table = instance.table("{BT_TABLE_TF}");

  const start = process.hrtime.bigint();
  const [row] = await table.row("A#000001").get({{
    filter: [{{column: {{cellLimit: 1}}}}],
  }});
  const elapsedMs = Number(process.hrtime.bigint() - start) / 1_000_000;

  console.log("=== Node.js client ===");
  console.log();
  console.log("Row key: A#000001");
  console.log(`Read latency: ${{elapsedMs.toFixed(1)}} ms`);
  console.log();

  for (const [family, columns] of Object.entries(row.data)) {{
    console.log(`Family: ${{family}}`);
    for (const [col, cells] of Object.entries(columns)) {{
      let val = cells[0].value;
      if (Buffer.isBuffer(val)) val = val.toString("utf8");
      if (val.length > 60) val = val.substring(0, 60) + "...";
      console.log(`  ${{col}} = ${{val}}`);
    }}
    console.log();
  }}
}}

main().catch(console.error);
'''

(node_dir / 'index.js').write_text(node_program)

# Initialize npm and install
print('Installing @google-cloud/bigtable for Node.js...')
subprocess.run(['npm', 'init', '-y'], cwd=str(node_dir), capture_output=True)
result = subprocess.run(
    ['npm', 'install', '--save', '@google-cloud/bigtable'],
    cwd=str(node_dir),
    capture_output=True, text=True,
    timeout=120,
)
if result.returncode != 0:
    print(f'npm install error: {result.stderr[:500]}')
else:
    print('Node.js dependencies installed.')
```

```python
print('Running Node.js client...\n')

result = subprocess.run(
    ['node', 'index.js'],
    cwd=str(node_dir),
    capture_output=True, text=True,
    timeout=30,
)
print(result.stdout)
if result.returncode != 0:
    print(f'Error: {result.stderr[:1000]}')
```

## 7. Client Comparison

All four clients read the same row from the same Bigtable instance. The latency differences come from client initialization overhead (first call vs warm connection), gRPC channel setup, and language runtime characteristics.

| Language | Client Library | Strengths | Best For |
|----------|---------------|-----------|----------|
| **Python** | `google-cloud-bigtable` | Ecosystem (pandas, sklearn), rapid prototyping | ML pipelines, notebooks, batch processing |
| **Go** | `cloud.google.com/go/bigtable` | Low latency, compiled binary, goroutines | High-throughput serving, microservices |
| **Java** | `com.google.cloud:google-cloud-bigtable` | Mature, enterprise-grade, strong typing | JVM-based serving, Spring Boot, Dataflow |
| **Node.js** | `@google-cloud/bigtable` | Async I/O, serverless, fast iteration | Cloud Functions, Cloud Run, API gateways |

### Production Considerations

- **Connection pooling**: all clients maintain internal gRPC channel pools. Create the client once at startup, reuse for all requests. See [Notebook 8 (Serving Integration)](./Bigtable%20Feature%20Store%20-%20Serving%20Integration.ipynb), Section 5 for Python connection pooling patterns — the same principle applies in every language.
- **App profiles**: pass the `serving` app profile to route reads through the high-priority path. The Terraform configuration in Section 2 created this profile.
- **Error handling**: all clients support automatic retries with exponential backoff for transient errors (UNAVAILABLE, DEADLINE_EXCEEDED). Configure timeouts based on your SLA.

## Cleanup

Terraform manages the full lifecycle — `terraform destroy` removes everything it created. No manual cleanup needed.

```python
print('Destroying Terraform-managed infrastructure...\n')

result = subprocess.run(
    [TF, 'destroy', '-auto-approve', '-no-color'],
    cwd=TF_DIR,
    capture_output=True, text=True,
    timeout=300,
)
print(result.stdout[-2000:] if len(result.stdout) > 2000 else result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr[-1000:])

# Verify instance is gone
try:
    instance.reload()
    print(f'WARNING: Instance {BT_INSTANCE_TF} still exists')
except Exception as e:
    print(f'Verified: instance {BT_INSTANCE_TF} has been deleted')

# Clean up local files
import shutil as sh
for d in [TF_DIR, 'go-bigtable-client', 'java-bigtable-client', 'node-bigtable-client']:
    p = pathlib.Path(d)
    if p.exists():
        sh.rmtree(p)
        print(f'Removed {d}/')

tf_bin = pathlib.Path('terraform')
if tf_bin.exists():
    tf_bin.unlink()
    print('Removed terraform binary')

mvn_dir = pathlib.Path('.').glob('apache-maven-*')
for d in mvn_dir:
    sh.rmtree(d)
    print(f'Removed {d}/')

print('\nCleanup complete.')
```
