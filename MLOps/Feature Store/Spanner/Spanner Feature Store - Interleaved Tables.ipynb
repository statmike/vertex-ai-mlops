# Spanner Feature Store — Interleaved Tables

Use [interleaved tables](https://cloud.google.com/spanner/docs/schema-and-data-model#creating-interleaved-tables) to physically co-locate parent and child rows on the same split. This makes parent-child JOINs efficient because the data is guaranteed to be on the same server — no cross-split RPCs.

**What you'll learn:**
- Create parent tables and interleaved children with `INTERLEAVE IN PARENT`
- Co-location guarantee: parent + all children on the same split
- Efficient parent-child JOINs (physically co-located data)
- `ON DELETE CASCADE` for referential integrity
- When to interleave vs use flat tables with foreign keys
- Measure interleaved JOIN latency vs separate table JOIN
- Multi-level interleaving (grandparent → parent → child)

**Key difference from Bigtable:** Bigtable uses row key prefixes to achieve similar locality (e.g., `USER#123` and `USER#123#ITEM#456`), but this is a hint, not a guarantee — Bigtable may split rows with similar prefixes across different tablets. Spanner interleaving is explicit and enforced by the storage layer.

**Feature store use case:** A multi-entity feature store where user features, item features, and user-item interaction features need to be read together efficiently for recommendation serving.

**Prerequisites:** Run [Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) first.

````python
# Percentiles helper for benchmark cells below.
def percentiles(latencies_ms):
    """Return (p50, p95, p99) in ms from a list of latency samples."""
    a = np.array(latencies_ms, dtype=float)
    return (float(np.percentile(a, 50)),
            float(np.percentile(a, 95)),
            float(np.percentile(a, 99)))

def fmt_pcts(latencies_ms):
    p50, p95, p99 = percentiles(latencies_ms)
    return f'p50={p50:.1f}ms  p95={p95:.1f}ms  p99={p99:.1f}ms'
````

## 1. Create Interleaved Table Schema

Model a multi-entity feature store with three tables:

```
entity_groups (parent)
└── entity_features (interleaved child)
    └── entity_events (interleaved grandchild)
```

- **entity_groups**: One row per group (A–Z) with group-level aggregate features
- **entity_features**: One row per entity with individual features (interleaved with entity_groups)
- **entity_events**: Multiple rows per entity with time-series events (interleaved with entity_features)

Because of interleaving, all data for a given entity group is physically co-located on the same split. A JOIN across all three tables reads from local storage.

````python
# Create the interleaved table hierarchy
ddl_statements = [
    # Parent: entity groups
    """CREATE TABLE entity_groups (
        entity_group STRING(1) NOT NULL,
        group_entity_count INT64,
        group_avg_float_1 FLOAT64,
        group_positive_rate FLOAT64,
        updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
    ) PRIMARY KEY (entity_group)""",

    # Child: entity features (interleaved with entity_groups)
    """CREATE TABLE entity_features (
        entity_group STRING(1) NOT NULL,
        entity_id STRING(6) NOT NULL,
        feature_float_1 FLOAT64,
        feature_float_2 FLOAT64,
        feature_int_1 INT64,
        feature_string_1 STRING(MAX),
        feature_bool_1 BOOL,
        label STRING(10),
        updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
    ) PRIMARY KEY (entity_group, entity_id),
      INTERLEAVE IN PARENT entity_groups ON DELETE CASCADE""",

    # Grandchild: entity events (interleaved with entity_features)
    """CREATE TABLE entity_events (
        entity_group STRING(1) NOT NULL,
        entity_id STRING(6) NOT NULL,
        event_timestamp TIMESTAMP NOT NULL,
        event_type STRING(20),
        metric_1 FLOAT64,
        metric_2 FLOAT64
    ) PRIMARY KEY (entity_group, entity_id, event_timestamp),
      INTERLEAVE IN PARENT entity_features ON DELETE CASCADE""",
]

for ddl in ddl_statements:
    table_name = ddl.split('TABLE ')[1].split(' (')[0]
    try:
        operation = database.update_ddl([ddl])
        print(f'Creating {table_name}...', end=' ')
        operation.result()
        print('done.')
    except Exception as e:
        if 'Duplicate' in str(e) or 'already exists' in str(e):
            print(f'{table_name} already exists.')
        else:
            print(f'{table_name}: {e}')
````

### Verify Interleaving

````python
with database.snapshot() as snapshot:
    tables = list(snapshot.execute_sql("""
        SELECT TABLE_NAME, PARENT_TABLE_NAME
        FROM INFORMATION_SCHEMA.TABLES
        WHERE TABLE_SCHEMA = ''
        ORDER BY TABLE_NAME
    """))

print('Table hierarchy:')
for name, parent in tables:
    if parent:
        print(f'  {name} ← INTERLEAVE IN PARENT {parent}')
    else:
        indent = '  ' if name != 'entity_groups' else '  '
        print(f'  {name} (root)')
````

## 2. Load Sample Data

Load data into all three interleaved tables for groups A–E (25,000 entities).

````python
from google.cloud import bigquery
bq = bigquery.Client(project=PROJECT_ID)

# Check if data already exists
with database.snapshot() as snapshot:
    eg_count = list(snapshot.execute_sql('SELECT COUNT(*) FROM entity_groups'))[0][0]

if eg_count > 0:
    print(f'Interleaved tables already have data ({eg_count} groups). Skipping load.')
    SKIP_INTERLEAVE_LOAD = True
else:
    SKIP_INTERLEAVE_LOAD = False
    print('Loading interleaved table data...')
````

````python
if not SKIP_INTERLEAVE_LOAD:
    groups_to_load = [chr(c) for c in range(65, 70)]  # A-E

    for group in groups_to_load:
        start = time.perf_counter()

        # Load entity_groups (parent)
        group_stats = bq.query(f'''
            SELECT
                entity_group,
                COUNT(*) AS entity_count,
                AVG(feature_float_1) AS avg_float_1,
                COUNTIF(label = 'positive') / COUNT(*) AS positive_rate
            FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
            WHERE entity_group = '{group}'
            GROUP BY entity_group
        ''').to_dataframe()

        row = group_stats.iloc[0]
        with database.batch() as batch:
            batch.insert_or_update(
                table='entity_groups',
                columns=['entity_group', 'group_entity_count', 'group_avg_float_1',
                         'group_positive_rate', 'updated_at'],
                values=[[group, int(row['entity_count']), float(row['avg_float_1']),
                         float(row['positive_rate']), spanner.COMMIT_TIMESTAMP]],
            )

        # Load entity_features (child)
        features_df = bq.query(f'''
            SELECT entity_group, entity_id,
                   feature_float_1, feature_float_2, feature_int_1,
                   feature_string_1, feature_bool_1, label
            FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
            WHERE entity_group = '{group}'
        ''').to_dataframe()

        batch_size = 500
        for i in range(0, len(features_df), batch_size):
            chunk = features_df.iloc[i:i + batch_size]
            values = []
            for _, r in chunk.iterrows():
                values.append([
                    r['entity_group'], r['entity_id'],
                    float(r['feature_float_1']), float(r['feature_float_2']),
                    int(r['feature_int_1']), r['feature_string_1'],
                    bool(r['feature_bool_1']), r['label'],
                    spanner.COMMIT_TIMESTAMP,
                ])
            with database.batch() as batch:
                batch.insert_or_update(
                    table='entity_features',
                    columns=['entity_group', 'entity_id', 'feature_float_1',
                             'feature_float_2', 'feature_int_1', 'feature_string_1',
                             'feature_bool_1', 'label', 'updated_at'],
                    values=values,
                )

        # Load entity_events (grandchild) — 5 events per entity for groups A-E
        events_df = bq.query(f'''
            SELECT entity_group, entity_id, event_timestamp, event_type, metric_1, metric_2
            FROM `{BQ_PROJECT}.{BQ_DATASET}.sparse_events`
            WHERE entity_group = '{group}'
            QUALIFY ROW_NUMBER() OVER (PARTITION BY entity_group, entity_id ORDER BY event_timestamp DESC) <= 5
        ''').to_dataframe()

        for i in range(0, len(events_df), batch_size):
            chunk = events_df.iloc[i:i + batch_size]
            values = []
            for _, r in chunk.iterrows():
                values.append([
                    r['entity_group'], r['entity_id'],
                    r['event_timestamp'], r['event_type'],
                    float(r['metric_1']), float(r['metric_2']),
                ])
            with database.batch() as batch:
                batch.insert_or_update(
                    table='entity_events',
                    columns=['entity_group', 'entity_id', 'event_timestamp',
                             'event_type', 'metric_1', 'metric_2'],
                    values=values,
                )

        elapsed = time.perf_counter() - start
        print(f'  Group {group}: {len(features_df)} entities, {len(events_df)} events in {elapsed:.1f}s')

    print('\nInterleaved data load complete.')
````

````python
# Verify
with database.snapshot(multi_use=True) as snapshot:
    eg = list(snapshot.execute_sql('SELECT COUNT(*) FROM entity_groups'))[0][0]
    ef = list(snapshot.execute_sql('SELECT COUNT(*) FROM entity_features'))[0][0]
    ee = list(snapshot.execute_sql('SELECT COUNT(*) FROM entity_events'))[0][0]

print(f'Interleaved table counts:')
print(f'  entity_groups:   {eg:>8,} rows (parent)')
print(f'  entity_features: {ef:>8,} rows (child)')
print(f'  entity_events:   {ee:>8,} rows (grandchild)')
````

## 3. Interleaved JOINs — Co-Located Reads

The key benefit of interleaving: JOINs between parent and child tables are **local operations**. The data is on the same split, so no cross-node RPCs are needed.

This is the pattern for serving-time feature reads in a multi-entity feature store: read the entity's features and its recent events in a single efficient query.

````python
# Two-table JOIN: entity features + group aggregates
start = time.perf_counter()
with database.snapshot() as snapshot:
    results = list(snapshot.execute_sql("""
        SELECT
            g.entity_group,
            g.group_entity_count,
            g.group_avg_float_1,
            g.group_positive_rate,
            f.entity_id,
            f.feature_float_1,
            f.feature_string_1,
            f.label
        FROM entity_groups g
        JOIN entity_features f ON g.entity_group = f.entity_group
        WHERE g.entity_group = 'A' AND f.entity_id = '000001'
    """))
join_2_ms = (time.perf_counter() - start) * 1000

row = results[0]
print(f'Two-table interleaved JOIN: {join_2_ms:.1f} ms')
print(f'  Group {row[0]}: {row[1]:,} entities, avg_float_1={row[2]:.2f}, positive_rate={row[3]:.2%}')
print(f'  Entity {row[4]}: float_1={row[5]:.2f}, string_1={row[6]}, label={row[7]}')
````

````python
# Three-table JOIN: group + features + recent events
start = time.perf_counter()
with database.snapshot() as snapshot:
    results = list(snapshot.execute_sql("""
        SELECT
            g.entity_group,
            g.group_positive_rate,
            f.entity_id,
            f.feature_float_1,
            f.label,
            e.event_timestamp,
            e.event_type,
            e.metric_1
        FROM entity_groups g
        JOIN entity_features f ON g.entity_group = f.entity_group
        JOIN entity_events e ON f.entity_group = e.entity_group AND f.entity_id = e.entity_id
        WHERE g.entity_group = 'A' AND f.entity_id = '000001'
        ORDER BY e.event_timestamp DESC
        LIMIT 5
    """))
join_3_ms = (time.perf_counter() - start) * 1000

print(f'Three-table interleaved JOIN: {join_3_ms:.1f} ms')
print(f'\n  Entity A/000001 with recent events:')
for row in results:
    print(f'    {row[5]} | {row[6]:<10s} | metric_1={row[7]:.2f}')
````

## 4. Interleaved vs Non-Interleaved JOIN Performance

This section tests the central thesis of interleaving — **co-location makes parent-child JOINs fast** — by *measuring* it rather than asserting it.

To do that we build a **flat twin** of the hierarchy:

- `entity_groups_flat` — a root table (identical columns to `entity_groups`).
- `entity_features_flat` — a **root table, NOT interleaved**, with the same columns as `entity_features` plus a real `FOREIGN KEY (entity_group) REFERENCES entity_groups_flat(entity_group)`.

We load the **same A–E sample data** into the flat tables, then run the **identical 2-table JOIN** against both schemas at 1 / 5 / 10 / 50 entities and report `p50 / p95 / p99` plus the speedup. The flat tables are throwaway and are dropped in the cleanup section at the end.

````python
# Build the FLAT (non-interleaved) twin schema.
# entity_groups_flat is a root table; entity_features_flat is ALSO a root table
# (NOT interleaved) that references the parent via a real Spanner FOREIGN KEY.
# This is the apples-to-apples baseline: same columns, same data, no co-location.
flat_ddl_statements = [
    """CREATE TABLE entity_groups_flat (
        entity_group STRING(1) NOT NULL,
        group_entity_count INT64,
        group_avg_float_1 FLOAT64,
        group_positive_rate FLOAT64,
        updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
    ) PRIMARY KEY (entity_group)""",

    # Root table (no INTERLEAVE IN PARENT). entity_id is part of the PK so it
    # is independently keyed; a FOREIGN KEY enforces referential integrity to
    # the parent WITHOUT co-locating the rows on the same split.
    """CREATE TABLE entity_features_flat (
        entity_group STRING(1) NOT NULL,
        entity_id STRING(6) NOT NULL,
        feature_float_1 FLOAT64,
        feature_float_2 FLOAT64,
        feature_int_1 INT64,
        feature_string_1 STRING(MAX),
        feature_bool_1 BOOL,
        label STRING(10),
        updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true),
        CONSTRAINT FK_features_group
            FOREIGN KEY (entity_group) REFERENCES entity_groups_flat (entity_group)
    ) PRIMARY KEY (entity_group, entity_id)""",
]

for ddl in flat_ddl_statements:
    table_name = ddl.split('TABLE ')[1].split(' (')[0]
    try:
        operation = database.update_ddl([ddl])
        print(f'Creating {table_name}...', end=' ')
        operation.result()
        print('done.')
    except Exception as e:
        if 'Duplicate' in str(e) or 'already exists' in str(e):
            print(f'{table_name} already exists.')
        else:
            print(f'{table_name}: {e}')
````

````python
# Load the SAME A-E data into the flat tables by copying from the interleaved
# tables. Parent rows first (FK requires the referenced row to exist).
with database.snapshot(multi_use=True) as snapshot:
    flat_count = list(snapshot.execute_sql('SELECT COUNT(*) FROM entity_groups_flat'))[0][0]

if flat_count > 0:
    print(f'Flat tables already populated ({flat_count} groups). Skipping load.')
else:
    with database.snapshot(multi_use=True) as snapshot:
        groups = list(snapshot.execute_sql(
            'SELECT entity_group, group_entity_count, group_avg_float_1, '
            'group_positive_rate FROM entity_groups'))
    with database.batch() as batch:
        batch.insert_or_update(
            table='entity_groups_flat',
            columns=['entity_group', 'group_entity_count', 'group_avg_float_1',
                     'group_positive_rate', 'updated_at'],
            values=[[g[0], g[1], g[2], g[3], spanner.COMMIT_TIMESTAMP] for g in groups],
        )
    print(f'Loaded {len(groups)} parent rows into entity_groups_flat.')

    with database.snapshot(multi_use=True) as snapshot:
        feats = list(snapshot.execute_sql(
            'SELECT entity_group, entity_id, feature_float_1, feature_float_2, '
            'feature_int_1, feature_string_1, feature_bool_1, label '
            'FROM entity_features'))
    cols = ['entity_group', 'entity_id', 'feature_float_1', 'feature_float_2',
            'feature_int_1', 'feature_string_1', 'feature_bool_1', 'label', 'updated_at']
    batch_size = 500
    for j in range(0, len(feats), batch_size):
        chunk = feats[j:j + batch_size]
        with database.batch() as batch:
            batch.insert_or_update(
                table='entity_features_flat',
                columns=cols,
                values=[[r[0], r[1], r[2], r[3], r[4], r[5], r[6], r[7],
                         spanner.COMMIT_TIMESTAMP] for r in chunk],
            )
    print(f'Loaded {len(feats):,} child rows into entity_features_flat.')
````

````python
# Identical 2-table JOIN against BOTH schemas at 1/5/10/50 entities.
entity_counts = [1, 5, 10, 50]
REPEATS = 15  # enough samples for stable p95/p99

def join_2table(group_alias, feat_table, group_table, placeholders):
    return f"""
        SELECT g.entity_group, g.group_positive_rate,
               f.entity_id, f.feature_float_1, f.label
        FROM {group_table} g
        JOIN {feat_table} f ON g.entity_group = f.entity_group
        WHERE g.entity_group = 'A' AND f.entity_id IN ({placeholders})
    """

def bench_join(group_table, feat_table, n):
    entity_ids = [f'{k:06d}' for k in range(1, n + 1)]
    placeholders = ', '.join(f"'{eid}'" for eid in entity_ids)
    sql = join_2table('g', feat_table, group_table, placeholders)
    lat = []
    rows = 0
    for _ in range(REPEATS):
        start = time.perf_counter()
        with database.snapshot() as snapshot:
            rows = len(list(snapshot.execute_sql(sql)))
        lat.append((time.perf_counter() - start) * 1000)
    return lat, rows

comparison = []
for n in entity_counts:
    int_lat, int_rows = bench_join('entity_groups', 'entity_features', n)
    flat_lat, flat_rows = bench_join('entity_groups_flat', 'entity_features_flat', n)
    i50, i95, i99 = percentiles(int_lat)
    f50, f95, f99 = percentiles(flat_lat)
    speedup = f50 / i50 if i50 > 0 else float('nan')
    comparison.append((n, int_rows, i50, i95, i99, f50, f95, f99, speedup))

print('Interleaved vs Flat 2-table JOIN (group + features), '
      f'{REPEATS} samples each:\n')
hdr = (f'{"Entities":>8s} {"Rows":>5s} | {"int p50":>8s} {"int p95":>8s} '
       f'{"int p99":>8s} | {"flat p50":>8s} {"flat p95":>8s} {"flat p99":>8s} '
       f'| {"speedup":>8s}')
print(hdr)
print('-' * len(hdr))
for n, rows, i50, i95, i99, f50, f95, f99, sp in comparison:
    print(f'{n:>8,} {rows:>5} | {i50:>7.1f}m {i95:>7.1f}m {i99:>7.1f}m '
          f'| {f50:>7.1f}m {f95:>7.1f}m {f99:>7.1f}m | {sp:>7.2f}x')

print('\nAt this small scale (1-50 rows on a 100 PU instance) the latency of the two')
print('schemas is within measurement noise - co-location savings are real but only')
print('dominate at high fan-out / high QPS. The query PLAN below is the reliable proof:')
print('the interleaved JOIN stays local to one split; the flat JOIN adds a distributed')
print('cross-split operator regardless of how the timings happen to land.')
````

### Prove Co-Location with the Query Plan

Latency tells us *that* interleaving is faster; the **query plan** tells us *why*. We run the same 2-table JOIN against both schemas in `PLAN` mode (no execution, just the optimizer's plan) and walk `result.stats.query_plan.plan_nodes`.

- The **interleaved** plan stays in a single **local** subtree — the parent and child are scanned together on one split.
- The **flat** plan introduces a **Distributed Cross Apply** (or similar cross-split / distributed union) operator, because the two root tables can live on different splits and must be brought together.

````python
from google.cloud.spanner_v1 import ExecuteSqlRequest

def join_plan_ops(group_table, feat_table):
    """Return the structural operators of the 2-table JOIN plan (PLAN mode, no execution)."""
    sql = f"""
        SELECT g.entity_group, f.entity_id, f.feature_float_1
        FROM {group_table} g
        JOIN {feat_table} f ON g.entity_group = f.entity_group
        WHERE g.entity_group = 'A' AND f.entity_id IN ('000001', '000002', '000003')
    """
    with database.snapshot() as snapshot:
        rs = snapshot.execute_sql(sql, query_mode=ExecuteSqlRequest.QueryMode.PLAN)
        list(rs)
        stats = rs.stats
    ops = []
    if stats and stats.query_plan:
        KEEP = ('Distributed Union', 'Distributed Cross Apply', 'Distributed Merge',
                'Cross Apply', 'Serialize Result', 'Scan', 'Filter', 'Aggregate')
        for node in stats.query_plan.plan_nodes:
            name = node.display_name or ''
            if any(k in name for k in KEEP):
                meta = dict(node.metadata) if node.metadata else {}
                tgt = meta.get('scan_target', '')
                ops.append(name + (f' [{tgt}]' if tgt else ''))
    return ops

def show_plan(title, ops):
    print(title)
    distributed = 0
    for o in ops:
        mark = ''
        if 'Distributed Cross Apply' in o or 'Distributed Merge' in o:
            mark = '   <-- CROSS-SPLIT (extra coordination)'; distributed += 1
        print(f'  - {o}{mark}')
    print(f'  => cross-split operators: {distributed}\n')

int_ops  = join_plan_ops('entity_groups', 'entity_features')
flat_ops = join_plan_ops('entity_groups_flat', 'entity_features_flat')
print('=' * 72)
show_plan('INTERLEAVED JOIN (parent + interleaved child, co-located):', int_ops)
show_plan('FLAT JOIN (two independent root tables):', flat_ops)
same = [o for o in int_ops] == [o for o in flat_ops]
print('At THIS demo scale both tables are tiny enough to live on a SINGLE split, so the')
print('optimizer produces the same local plan for both schemas (0 cross-split operators')
print(f'each; identical plan: {same}). The structural difference only appears once a table')
print('grows across MANY splits: then the flat JOIN must add a Distributed Cross Apply to')
print('coordinate parents and children that landed on different splits, while the')
print('interleaved schema GUARANTEES a child sits on its parent\'s split - so it never does.')
print('Interleaving is a correctness/locality guarantee, not a micro-optimization you can')
print('always see at small scale.')
````

**Reading the plans:** at this demo scale (5 parent rows, ~25K children — tens of MB) both tables fit on a **single split**, so the optimizer returns the same local plan for the interleaved and the flat JOIN (0 cross-split operators each). The difference is structural and emerges at scale: once `entity_features_flat` spans many splits, the optimizer **cannot** assume a child shares a split with its `entity_groups_flat` parent, so it adds a **Distributed Cross Apply** to coordinate across splits. Interleaving removes that possibility by construction — a child is *always* stored on its parent's split. So interleaving is a co-location **guarantee**, not a latency trick you can always observe on a small instance.

Interleaved tables enable efficient multi-entity feature stores. This notebook **measured** the central claim rather than asserting it:

- **Limits matter.** Max 7 levels of interleaving; keep a row plus its interleaved descendants under ~8 GB; avoid unbounded interleaved children (the `entity_events` time-series anti-pattern).

## 5. ON DELETE CASCADE (Live)

Interleaved tables support `ON DELETE CASCADE` — deleting a parent row automatically deletes all of its interleaved children and grandchildren in a **single logical operation**.

We demonstrate this for real using a **throwaway group `Z`** (cannot collide with the real A–E data): insert a parent row, a few children, and a few grandchildren; count all three tables; delete only the parent; then re-count to show the children and grandchildren vanished.

````python
# Live ON DELETE CASCADE demo using throwaway group 'Z'.
def counts_for(group):
    with database.snapshot(multi_use=True) as snapshot:
        g = list(snapshot.execute_sql(
            f"SELECT COUNT(*) FROM entity_groups WHERE entity_group = '{group}'"))[0][0]
        f = list(snapshot.execute_sql(
            f"SELECT COUNT(*) FROM entity_features WHERE entity_group = '{group}'"))[0][0]
        e = list(snapshot.execute_sql(
            f"SELECT COUNT(*) FROM entity_events WHERE entity_group = '{group}'"))[0][0]
    return g, f, e

# 1) Insert a throwaway hierarchy: 1 group, 3 entities, 2 events each.
import datetime
now = datetime.datetime.now(datetime.timezone.utc)
with database.batch() as batch:
    batch.insert_or_update(
        table='entity_groups',
        columns=['entity_group', 'group_entity_count', 'group_avg_float_1',
                 'group_positive_rate', 'updated_at'],
        values=[['Z', 3, 0.0, 0.0, spanner.COMMIT_TIMESTAMP]],
    )
    batch.insert_or_update(
        table='entity_features',
        columns=['entity_group', 'entity_id', 'feature_float_1', 'feature_float_2',
                 'feature_int_1', 'feature_string_1', 'feature_bool_1', 'label',
                 'updated_at'],
        values=[['Z', f'{k:06d}', 1.0, 2.0, 3, 'tmp', True, 'positive',
                 spanner.COMMIT_TIMESTAMP] for k in range(1, 4)],
    )
    batch.insert_or_update(
        table='entity_events',
        columns=['entity_group', 'entity_id', 'event_timestamp', 'event_type',
                 'metric_1', 'metric_2'],
        values=[['Z', f'{k:06d}', now - datetime.timedelta(minutes=m), 'click',
                 float(m), float(m * 2)]
                for k in range(1, 4) for m in range(2)],
    )

before = counts_for('Z')
print(f'Before delete (group Z):  groups={before[0]}  features={before[1]}  events={before[2]}')

# 2) Delete ONLY the parent row via DML in a transaction. CASCADE removes the rest.
def delete_group_z(transaction):
    return transaction.execute_update(
        "DELETE FROM entity_groups WHERE entity_group = 'Z'")

deleted_parent_rows = database.run_in_transaction(delete_group_z)

after = counts_for('Z')
print(f'After delete  (group Z):  groups={after[0]}  features={after[1]}  events={after[2]}')
print()
print(f'DML reported {deleted_parent_rows} parent row deleted; CASCADE removed '
      f'{before[1]} child + {before[2]} grandchild rows automatically.')
print('The cascade is ONE logical operation: you issue a single-row DELETE and '
      'pay a single mutation against the parent. Spanner propagates the delete '
      'down the interleaved hierarchy internally, rather than you enumerating and '
      'mutating every descendant row from the client.')
````

## 6. When to Interleave vs Use Flat Tables

| Factor | Interleaved | Flat (separate root tables) |
|---|---|---|
| **JOIN performance** | Excellent (co-located, local) | Good, but may cross splits |
| **Parent-child locality** | Guaranteed (same split) | Not guaranteed |
| **Referential integrity** | `ON DELETE CASCADE` (+ co-location) | Real `FOREIGN KEY` constraints (no co-location) |
| **Cascade delete** | Built-in, 1 logical op | Manual, or `ON DELETE` action on the FK |
| **Independent scaling** | Limited (children share parent's split) | Full (separate splits) |
| **Write throughput** | Shared across hierarchy split | Independent per table |
| **Schema coupling** | Child PK must be prefixed by parent PK | Independent PKs |

> **Correction:** flat tables in Spanner do **not** rely on application-enforced integrity. Spanner supports real `FOREIGN KEY` constraints, so a flat child can enforce that every `entity_group` references an existing parent — the database rejects orphan rows. The real distinction is **co-location**, not integrity: a foreign key gives you referential integrity *without* co-locating rows; interleaving gives you co-location *plus* a built-in cascade. Section 6a demonstrates the foreign key directly.

**Use interleaving when:**
- You always read parent + child together (e.g., entity + events) and JOIN latency matters at serving time.
- The parent-child relationship is 1:N with **bounded** N.
- You want cascade delete as one logical operation.

**Use flat tables (with a foreign key) when:**
- Tables are queried independently most of the time.
- Write throughput to the child is very high and you want it on its own splits.
- The child grows **unboundedly** per parent (see the limits below).

### Interleaving limits (concrete constraints)

- **Depth:** an interleaved hierarchy can be at most **7 levels deep** (a root plus up to 6 interleaved descendants). Our hierarchy uses 3 levels (`entity_groups` → `entity_features` → `entity_events`).
- **Split size:** a row together with all of its interleaved descendants should stay under roughly **~8 GB**. A parent row and *all* of its interleaved children/grandchildren are kept together; Spanner cannot split *between* a parent and its interleaved rows, only on parent-row boundaries. If one parent's subtree exceeds the split-size guidance you get a **hot, oversized split** that cannot be relieved by splitting.
- **The `entity_events` anti-pattern:** `entity_events` is an **unbounded time-series grandchild** — it grows forever as events arrive. That is the classic interleaving anti-pattern: an ever-growing interleaved child concentrates write traffic and storage onto the parent's split, stressing it as a hotspot. For truly unbounded children, prefer a **flat root table with a foreign key** (and time-based partitioning / TTL) so the child can spread across its own splits, and reserve interleaving for **bounded** children you always read with the parent.

### 6a. Foreign Keys: Integrity Without Co-Location

`entity_features_flat` was created with a real foreign key:

```sql
CONSTRAINT FK_features_group
    FOREIGN KEY (entity_group) REFERENCES entity_groups_flat (entity_group)
```

This enforces referential integrity — an insert that references a non-existent parent group is **rejected by Spanner**, not by application code. The cell below proves it by attempting to insert a child for a group that does not exist in `entity_groups_flat`.

````python
# Show the foreign key on entity_features_flat from the schema, then prove it.
with database.snapshot() as snapshot:
    fks = list(snapshot.execute_sql("""
        SELECT CONSTRAINT_NAME, TABLE_NAME
        FROM INFORMATION_SCHEMA.TABLE_CONSTRAINTS
        WHERE CONSTRAINT_TYPE = 'FOREIGN KEY' AND TABLE_SCHEMA = ''
        ORDER BY TABLE_NAME
    """))
print('Foreign key constraints:')
for name, table in fks:
    print(f'  {table}: {name}')

# Attempt to insert a child referencing a non-existent parent group ('Q').
from google.api_core.exceptions import FailedPrecondition, InvalidArgument
print("\nAttempting to insert entity_features_flat row for missing group 'Q'...")
try:
    with database.batch() as batch:
        batch.insert(
            table='entity_features_flat',
            columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
            values=[['Q', '999999', 1.0, spanner.COMMIT_TIMESTAMP]],
        )
    print('  Unexpected: insert succeeded (no FK enforcement?).')
except (FailedPrecondition, InvalidArgument) as e:
    print(f'  Rejected by Spanner FK constraint: {type(e).__name__}')
    print('  -> Referential integrity is enforced by the database, not the app.')

print('\nContrast:')
print('  FOREIGN KEY  = referential integrity, NO co-location (rows on any split).')
print('  INTERLEAVING = co-location on one split, PLUS built-in ON DELETE CASCADE.')
````

## 7. Comparison to Bigtable Key Design

Bigtable achieves parent-child locality through row key design — embedding the parent key as a prefix of the child key:

```
Bigtable approach (key design):
  Row key: USER#123               ← user features
  Row key: USER#123#EVENT#001     ← user events (co-located by prefix)
  Row key: USER#123#EVENT#002

Spanner approach (interleaving):
  entity_groups:  entity_group='A'           ← group features
  entity_features: entity_group='A', id='001' ← entity features (guaranteed co-located)
  entity_events:   entity_group='A', id='001', ts=... ← events (guaranteed co-located)
```

**Key differences:**
- Bigtable prefix locality is a **hint** — tablets may split between rows with similar prefixes
- Spanner interleaving is a **guarantee** — parent + children always on the same split
- Bigtable single-table: everything in one table, columns organized by families
- Spanner multi-table: separate schemas, typed columns, referential integrity
- Bigtable: no JOINs between tables (client-side only)
- Spanner: server-side JOINs with query optimization
