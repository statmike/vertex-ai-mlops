# Spanner Feature Store — Emulator

Local development with the [Spanner emulator](https://cloud.google.com/spanner/docs/emulator). Test your feature store code without a GCP project or billing.

**What you'll learn:**
- Start the Spanner emulator: `gcloud emulators spanner start`
- `SPANNER_EMULATOR_HOST` environment variable for automatic client routing
- Same Python client code works against emulator and production
- What the emulator supports vs what it doesn't
- pytest fixtures for emulator-backed tests
- CI/CD patterns with GitHub Actions and Cloud Build

**Key difference from Bigtable:** The Bigtable emulator (`gcloud beta emulators bigtable start`) has similar limitations. The Spanner emulator supports full SQL, transactions, and DDL — making it more useful for testing feature store code.

**Prerequisites:** `gcloud` CLI installed.

---
## 1. Start the Emulator

The Spanner emulator runs locally and accepts the same API calls as production Spanner.

**Ports.** The emulator exposes two endpoints:

| Port | Protocol | Used by |
|---|---|---|
| `9010` | gRPC | The Python client (`SPANNER_EMULATOR_HOST=localhost:9010`) |
| `9020` | REST / HTTP proxy | `gcloud` and any REST-based tooling (`localhost:9020`) |

```bash
# Start the emulator (gRPC on 9010, REST proxy on 9020)
gcloud emulators spanner start

# In another terminal, set the environment variable (gRPC endpoint)
export SPANNER_EMULATOR_HOST=localhost:9010
```

When `SPANNER_EMULATOR_HOST` is set, the Python client automatically routes to the emulator instead of production Spanner. The GitHub Actions service container and Cloud Build example later in this notebook map **both** ports to stay consistent with this setup.

````python
import subprocess, sys, shutil

def install(*packages):
    uv = shutil.which('uv')
    if uv:
        subprocess.check_call([uv, 'pip', 'install', '-q', '--prerelease=allow', '--python', sys.executable, *packages])
    else:
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', '--upgrade', *packages])

install('google-cloud-spanner')
````

````python
import os
import time
import socket
import subprocess

# The emulator exposes gRPC on 9010 (client traffic) and a REST proxy on 9020.
EMULATOR_GRPC_PORT = 9010
EMULATOR_REST_PORT = 9020
EMULATOR_HOST = f'localhost:{EMULATOR_GRPC_PORT}'
READINESS_TIMEOUT_S = 30  # never hang forever

# Set in this cell so EVERY downstream cell can gate on it. Critically, we only
# point SPANNER_EMULATOR_HOST at the emulator once it is confirmed listening —
# otherwise the Spanner client would try to connect to a dead port and hang
# indefinitely (gRPC has no default deadline on connect).
EMULATOR_AVAILABLE = False


def _port_open(host, port, timeout=1.0):
    """Return True if a TCP connection to host:port succeeds."""
    try:
        with socket.create_connection((host, port), timeout=timeout):
            return True
    except OSError:
        return False


def wait_for_emulator(host='localhost', port=EMULATOR_GRPC_PORT, timeout=READINESS_TIMEOUT_S):
    """Poll the gRPC port until it accepts connections or we time out.

    This replaces a fixed `time.sleep()` — startup time varies, and a fixed
    wait is the most common cause of flaky Restart-&-Run-All failures.
    """
    deadline = time.time() + timeout
    while time.time() < deadline:
        if _port_open(host, port):
            return True
        time.sleep(0.5)
    return False


emulator_proc = None
try:
    emulator_proc = subprocess.Popen(
        ['gcloud', 'emulators', 'spanner', 'start'],
        stdout=subprocess.PIPE, stderr=subprocess.PIPE,
    )
    if wait_for_emulator():
        os.environ['SPANNER_EMULATOR_HOST'] = EMULATOR_HOST
        EMULATOR_AVAILABLE = True
        print(f'Emulator ready at {EMULATOR_HOST} (gRPC) / localhost:{EMULATOR_REST_PORT} (REST)')
        print(f'SPANNER_EMULATOR_HOST={os.environ["SPANNER_EMULATOR_HOST"]}')
    else:
        # Do NOT set SPANNER_EMULATOR_HOST — a client pointed at a dead port hangs.
        err = b''
        if emulator_proc and emulator_proc.poll() is not None and emulator_proc.stderr:
            err = emulator_proc.stderr.read()[:300]
        print(f'Emulator did not become ready within {READINESS_TIMEOUT_S}s — skipping live cells.')
        if err:
            print(f'  stderr: {err.decode(errors="replace")}')
        print('  Install it with: gcloud components install cloud-spanner-emulator')
        print('  (or run it via Docker: gcr.io/cloud-spanner-emulator/emulator)')
        if emulator_proc and emulator_proc.poll() is None:
            emulator_proc.terminate()  # don't leave a zombie if it half-started
except FileNotFoundError:
    print('gcloud CLI not found — emulator demo requires gcloud or the Docker image.')
    print('Install: https://cloud.google.com/sdk/docs/install')
    print('Continuing with conceptual examples (live cells will be skipped).')

print(f'\nEMULATOR_AVAILABLE = {EMULATOR_AVAILABLE}')
````

## 2. Create Resources on the Emulator

The emulator requires the same setup sequence as production: create instance → create database → create tables. The API is identical.

````python
import os  # noqa
# Silence the Spanner client's built-in Cloud Monitoring metrics exporter
# (noisy on incomplete resource labels; unrelated to the notebook).
os.environ.setdefault("SPANNER_DISABLE_BUILTIN_METRICS", "true")
from google.cloud import spanner

EMU_PROJECT = 'emulator-project'
EMU_INSTANCE = 'emulator-instance'
EMU_DATABASE = 'test-features'

emu_db = None

if not EMULATOR_AVAILABLE:
    # Guard: with no emulator listening, a spanner.Client connect would hang
    # indefinitely on the dead port. Skip all live cells cleanly.
    print('Emulator not running — skipping live emulator cells.')
    print('Install with: gcloud components install cloud-spanner-emulator')
    print('(or run the Docker image: gcr.io/cloud-spanner-emulator/emulator)')
else:
    try:
        emu_client = spanner.Client(project=EMU_PROJECT)

        # NOTE: the emulator accepts an ARBITRARY instance-config name — it does
        # not validate against real Spanner configs. If this fails, the cause is
        # the emulator being unreachable, not a bad config name.
        emu_instance = emu_client.instance(
            EMU_INSTANCE,
            configuration_name=f'projects/{EMU_PROJECT}/instanceConfigs/emulator-config',
            display_name='Emulator Test',
        )
        if not emu_instance.exists():
            emu_instance.create().result()
            print(f'Created emulator instance: {EMU_INSTANCE}')
        else:
            print(f'Emulator instance exists: {EMU_INSTANCE}')

        ddl = [
            """CREATE TABLE features (
                entity_group STRING(1) NOT NULL,
                entity_id STRING(6) NOT NULL,
                feature_float_1 FLOAT64,
                feature_string_1 STRING(MAX),
                feature_bool_1 BOOL,
                label STRING(10),
                updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
            ) PRIMARY KEY (entity_group, entity_id)""",
        ]
        emu_db = emu_instance.database(EMU_DATABASE, ddl_statements=ddl)
        if not emu_db.exists():
            emu_db.create().result()
            print(f'Created emulator database: {EMU_DATABASE}')
        else:
            # The db already exists (e.g. an emulator left running from a prior
            # run on port 9010). Drop and recreate so the schema is exactly the
            # one this notebook expects — otherwise later writes hit "column not
            # found" against a stale schema.
            emu_db.drop()
            emu_db = emu_instance.database(EMU_DATABASE, ddl_statements=ddl)
            emu_db.create().result()
            print(f'Recreated emulator database: {EMU_DATABASE} (was stale)')
    except Exception as e:
        print(f'Emulator not available: {e}')
        EMULATOR_AVAILABLE = False
````

````python
if EMULATOR_AVAILABLE:
    # Write and read data — same code as production
    with emu_db.batch() as batch:
        batch.insert_or_update(
            table='features',
            columns=['entity_group', 'entity_id', 'feature_float_1',
                     'feature_string_1', 'feature_bool_1', 'label', 'updated_at'],
            values=[
                ['A', '000001', 42.5, 'cat_1', True, 'positive', spanner.COMMIT_TIMESTAMP],
                ['A', '000002', 88.3, 'cat_2', False, 'negative', spanner.COMMIT_TIMESTAMP],
                ['B', '000001', 15.7, 'cat_3', True, 'negative', spanner.COMMIT_TIMESTAMP],
            ],
        )

    with emu_db.snapshot() as snapshot:
        results = list(snapshot.execute_sql("""
            SELECT entity_group, entity_id, feature_float_1, label
            FROM features
            ORDER BY entity_group, entity_id
        """))

    print('Data in emulator:')
    for row in results:
        print(f'  {row[0]}/{row[1]}: float_1={row[2]}, label={row[3]}')

    print(f'\nSame code, same API — just targeting the emulator instead of production.')
````

## 3. Emulator Capabilities and Limitations

| Capability | Emulator | Production |
|---|---|---|
| SQL (CRUD, DDL) | ✔ | ✔ |
| Transactions (ACID) | ✔ | ✔ |
| Secondary indexes | ✔ | ✔ |
| Interleaved tables | ✔ | ✔ |
| Change streams | ✔ | ✔ |
| INFORMATION_SCHEMA | ✔ | ✔ |
| Native vector search (`COSINE_DISTANCE`, `CREATE VECTOR INDEX`, `APPROX_*`) | verify at run time ↓ | ✔ |
| Multi-region | ✘ | ✔ |
| Stale reads | ✘ (always strong) | ✔ |
| IAM / fine-grained access | ✘ | ✔ |
| Performance benchmarking | ✘ (not representative) | ✔ |
| Persistence across restarts | ✘ (in-memory) | ✔ |
| Query optimizer behavior | Simplified | Production-grade |

**Bottom line:** The emulator is excellent for testing schema design, SQL queries, transactions, and application logic. It's not suitable for performance benchmarking or multi-region testing.

> **Vector search — verify at run time.** The [Vector Embeddings notebook](./Spanner%20Feature%20Store%20-%20Vector%20Embeddings.ipynb) uses native `COSINE_DISTANCE`, `CREATE VECTOR INDEX`, and `APPROX_COSINE_DISTANCE`. Emulator support for these varies by emulator version, so the next cell **probes the running emulator** rather than asserting a fixed answer.

````python
# Vector-search capability probe — VERIFY AT RUN TIME.
# Emulator support for native vector search varies by version, so we test each
# feature against the running emulator instead of assuming an answer.
import os

# Used by pytest skipif markers below: True when routed at the emulator.
ON_EMULATOR = 'SPANNER_EMULATOR_HOST' in os.environ

vector_support = {
    'COSINE_DISTANCE (scalar)': None,
    'CREATE VECTOR INDEX': None,
    'APPROX_COSINE_DISTANCE': None,
}

# JSON option string for the ANN query (kept separate to avoid nested quoting).
ANN_OPTS = '{"num_leaves_to_search": 10}'

if EMULATOR_AVAILABLE:
    # (a) COSINE_DISTANCE scalar. Use ARRAY<FLOAT32>[...] literals — you cannot
    #     CAST a FLOAT64 array literal to ARRAY<FLOAT32>.
    try:
        with emu_db.snapshot() as snapshot:
            list(snapshot.execute_sql(
                'SELECT COSINE_DISTANCE(ARRAY<FLOAT32>[1.0, 2.0, 3.0], '
                'ARRAY<FLOAT32>[4.0, 5.0, 6.0])'))
        vector_support['COSINE_DISTANCE (scalar)'] = True
    except Exception as e:
        vector_support['COSINE_DISTANCE (scalar)'] = False
        print(f'COSINE_DISTANCE unsupported: {str(e)[:120]}')

    # (b) CREATE VECTOR INDEX (DDL). The indexed embedding column must be NOT NULL
    #     (or the index NULL-filtered). Probe with a throwaway table.
    try:
        emu_db.update_ddl([
            'CREATE TABLE vec_probe (id STRING(36) NOT NULL, '
            'embedding ARRAY<FLOAT32>(vector_length=>3) NOT NULL) PRIMARY KEY (id)'
        ]).result()
        emu_db.update_ddl([
            'CREATE VECTOR INDEX vec_probe_idx ON vec_probe(embedding) '
            "OPTIONS (distance_type = 'COSINE')"
        ]).result()
        vector_support['CREATE VECTOR INDEX'] = True
    except Exception as e:
        vector_support['CREATE VECTOR INDEX'] = False
        print(f'CREATE VECTOR INDEX unsupported: {str(e)[:120]}')

    # (c) APPROX_COSINE_DISTANCE — only valid as the ORDER BY key of a KNN query
    #     against a vector-indexed table, not as a standalone scalar SELECT.
    if vector_support['CREATE VECTOR INDEX']:
        approx_sql = (
            'SELECT id FROM vec_probe ORDER BY APPROX_COSINE_DISTANCE('
            'embedding, ARRAY<FLOAT32>[1.0, 2.0, 3.0], '
            "options => JSON '" + ANN_OPTS + "') LIMIT 1"
        )
        try:
            with emu_db.snapshot() as snapshot:
                list(snapshot.execute_sql(approx_sql))
            vector_support['APPROX_COSINE_DISTANCE'] = True
        except Exception as e:
            vector_support['APPROX_COSINE_DISTANCE'] = False
            print(f'APPROX_COSINE_DISTANCE unsupported: {str(e)[:120]}')

    # Best-effort cleanup of probe objects.
    for stmt in ('DROP INDEX vec_probe_idx', 'DROP TABLE vec_probe'):
        try:
            emu_db.update_ddl([stmt]).result()
        except Exception:
            pass
else:
    print('Emulator not available — skipping vector-search probe.')

print('\nVector-search support on THIS emulator (verify at run time):')
for feature, supported in vector_support.items():
    status = 'not tested' if supported is None else ('supported' if supported else 'UNSUPPORTED')
    print(f'  {feature:32s}: {status}')

ON_EMULATOR_msg = 'routed at emulator' if ON_EMULATOR else 'routed at production'
print(f'\nON_EMULATOR = {ON_EMULATOR} ({ON_EMULATOR_msg})')
````

### In Practice: Feature-Store Testing on the Emulator

What you can confidently test locally vs what needs a real Spanner instance:

| You CAN test on the emulator | You CANNOT test on the emulator |
|---|---|
| Transactional multi-table feature updates (`run_in_transaction`) | Multi-region / multi-config placement and failover |
| `JOIN`s across feature tables | Fine-grained IAM / database roles / row-level access |
| Secondary indexes (creation + query planning correctness) | Performance, latency, and QPS (not representative) |
| DDL: `CREATE`/`ALTER`/`DROP TABLE`, interleaved tables, change streams | Stale / bounded-staleness reads (emulator is always strong) |
| Application logic, schema design, query correctness | Native vector indexes — **verify at run time** (see probe above) |

For anything in the right column, run the test against a dedicated production test instance instead. The skipif pattern below lets one test suite serve both.

```python
import os
import pytest

# True whenever the client is routed to the emulator.
ON_EMULATOR = 'SPANNER_EMULATOR_HOST' in os.environ

@pytest.mark.skipif(
    ON_EMULATOR,
    reason='Multi-region placement / IAM / vector indexes may be unsupported on the emulator',
)
def test_vector_index_knn():
    # Runs only against a real Spanner instance.
    ...

def test_transactional_feature_update():
    # Works on both emulator and production.
    ...
```

### pytest Fixture

````python
pytest_fixture_code = '''
# conftest.py
import os
import pytest
from google.cloud import spanner

EMULATOR_HOST = os.environ.get('SPANNER_EMULATOR_HOST', 'localhost:9010')

@pytest.fixture(scope='session')
def spanner_db():
    """Create a test database on the Spanner emulator."""
    os.environ['SPANNER_EMULATOR_HOST'] = EMULATOR_HOST
    client = spanner.Client(project='test-project')
    instance = client.instance('test-instance',
        configuration_name='projects/test-project/instanceConfigs/emulator-config')
    if not instance.exists():
        instance.create().result()

    ddl = [
        """CREATE TABLE features (
            entity_group STRING(1) NOT NULL,
            entity_id STRING(6) NOT NULL,
            feature_float_1 FLOAT64,
            updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
        ) PRIMARY KEY (entity_group, entity_id)"""
    ]
    db = instance.database('test-db', ddl_statements=ddl)
    if not db.exists():
        db.create().result()
    yield db

    # Cleanup
    db.drop()


# test_features.py
def test_write_and_read(spanner_db):
    with spanner_db.batch() as batch:
        batch.insert('features',
            columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
            values=[['A', '000001', 42.5, spanner.COMMIT_TIMESTAMP]])

    with spanner_db.snapshot() as snapshot:
        result = list(snapshot.execute_sql(
            "SELECT feature_float_1 FROM features WHERE entity_group = 'A' AND entity_id = '000001'"
        ))
    assert result[0][0] == 42.5


def test_transaction_atomicity(spanner_db):
    def atomic_update(transaction):
        result = list(transaction.execute_sql(
            "SELECT feature_float_1 FROM features WHERE entity_group = 'A' AND entity_id = '000001'"
        ))
        transaction.update('features',
            columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
            values=[['A', '000001', result[0][0] + 1.0, spanner.COMMIT_TIMESTAMP]])

    spanner_db.run_in_transaction(atomic_update)
    with spanner_db.snapshot() as snapshot:
        result = list(snapshot.execute_sql(
            "SELECT feature_float_1 FROM features WHERE entity_group = 'A' AND entity_id = '000001'"
        ))
    assert result[0][0] == 43.5
'''

print('pytest fixture for Spanner emulator testing:')
print(pytest_fixture_code)
````

### Test-Suite Runtime (not latency)

Per-operation **latency** on the emulator is meaningless (it's a local in-memory process — see the limitations table). What *is* useful for CI planning is the **wall-clock time of a small test suite**: if a representative batch of feature-store operations runs in well under a second, the emulator is cheap enough to run on every push.

The cell below times ~10 quick operations (writes + transactional update + reads) against the running emulator.

````python
# Time a small suite of operations to motivate the emulator for CI.
# NOTE: this measures SUITE RUNTIME, not latency — emulator latency is not
# representative of production and should not be used for benchmarking.
import time

if EMULATOR_AVAILABLE:
    N = 10
    t0 = time.perf_counter()
    for i in range(N):
        # write
        with emu_db.batch() as batch:
            batch.insert_or_update(
                table='features',
                columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
                values=[['T', str(i).zfill(6), float(i), spanner.COMMIT_TIMESTAMP]],
            )
        # transactional read-modify-write
        def _bump(txn, key=str(i).zfill(6)):
            rows = list(txn.execute_sql(
                "SELECT feature_float_1 FROM features "
                "WHERE entity_group = 'T' AND entity_id = @id",
                params={'id': key}, param_types={'id': spanner.param_types.STRING},
            ))
            txn.update(
                'features',
                columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
                values=[['T', key, rows[0][0] + 1.0, spanner.COMMIT_TIMESTAMP]],
            )
        emu_db.run_in_transaction(_bump)
        # read
        with emu_db.snapshot() as snapshot:
            list(snapshot.execute_sql(
                "SELECT feature_float_1 FROM features WHERE entity_group = 'T'"
            ))
    elapsed = time.perf_counter() - t0

    ops = N * 3  # write + txn + read per iteration
    print(f'Ran {ops} operations across {N} iterations in {elapsed:.3f}s')
    print(f'  ~{elapsed / ops * 1000:.1f} ms/op of SUITE time (NOT a latency benchmark)')
    print('  Fast enough to run on every push in CI.')

    # Clean up the timing rows.
    with emu_db.batch() as batch:
        for i in range(N):
            batch.delete(
                'features',
                spanner.KeySet(keys=[['T', str(i).zfill(6)]]),
            )
else:
    print('Emulator not available — skipping runtime measurement.')
````


### CI/CD with Cloud Build

The title of this notebook promises both GitHub Actions *and* Cloud Build. Cloud Build has no native "service container" for the emulator, so run it as a **background step** before the test step. Map both ports (`9010` gRPC, `9020` REST) for consistency with the GitHub Actions setup above.

```yaml
# cloudbuild.yaml
steps:
  # Start the Spanner emulator in the background (gRPC 9010 / REST 9020).
  - id: start-emulator
    name: gcr.io/cloud-spanner-emulator/emulator
    entrypoint: bash
    args:
      - -c
      - |
        ./gateway_main --hostname 0.0.0.0 \
          --grpc_port 9010 --http_port 9020 &
        echo "emulator started"
    waitFor: ['-']

  # Run the test suite against the emulator.
  - id: pytest
    name: python:3.13
    entrypoint: bash
    env:
      - 'SPANNER_EMULATOR_HOST=localhost:9010'
    args:
      - -c
      - |
        pip install -q google-cloud-spanner pytest
        # Wait for the gRPC port instead of a fixed sleep (mirrors the notebook).
        for i in $(seq 1 30); do
          python -c "import socket; socket.create_connection(('localhost',9010),1)" \
            2>/dev/null && break
          sleep 1
        done
        pytest tests/ -v
    waitFor: ['start-emulator']

options:
  # Steps share a network namespace so localhost:9010 reaches the emulator.
  pool: {}
```

> Cloud Build runs each step in its own container but on a shared Docker network, so the `pytest` step reaches the background emulator at `localhost:9010`. The readiness loop in the script mirrors the notebook's startup loop — no fixed `sleep` that could flake.
