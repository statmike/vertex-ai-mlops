# Spanner Feature Store — Environment

This notebook creates the shared environment for the Spanner Feature Store series. It verifies the BigQuery dataset (for the Spanner Feature Store series) and provisions a [Cloud Spanner](https://cloud.google.com/spanner/docs/overview) instance for online serving.

**What this notebook creates:**
- BigQuery dataset with two tables (same entity structure as the other feature store series):
  - **Dense table** (~130,000 rows) — 1 row per entity, 200+ features covering every BQ data type
  - **Sparse table** (~2.6M rows) — multiple rows per entity with timestamps for time-series features
- Cloud Spanner instance, database, and feature table with composite primary key

**Compound entity key:** All data uses a two-part key (`entity_group` + `entity_id`). In Spanner, this maps directly to a composite primary key `(entity_group, entity_id)` — no key concatenation needed.

### Create Tables (if missing)

This generates the 130K-entity dataset for the Spanner Feature Store series.

````python
if 'dense_features' in missing:
    print('Creating dense_features table (130K rows, 224 columns)...')

    float_columns = []
    for i in range(6, 181):
        if i % 5 == 0:
            float_columns.append(f'    ROUND(RAND() * 1000, 4) AS feature_float_{i}')
        elif i % 5 == 1:
            float_columns.append(f'    ROUND(RAND() * 100 - 50, 4) AS feature_float_{i}')
        elif i % 5 == 2:
            float_columns.append(f'    ROUND(RAND() * 10, 4) AS feature_float_{i}')
        elif i % 5 == 3:
            float_columns.append(f'    ROUND(RAND() * 500 + 100, 4) AS feature_float_{i}')
        else:
            float_columns.append(f'    ROUND(RAND(), 6) AS feature_float_{i}')

    float_cols_sql = ',\n'.join(float_columns)

    dense_query = f"""
    CREATE OR REPLACE TABLE `{BQ_PROJECT}.{BQ_DATASET}.dense_features` AS
    WITH
    entity_groups AS (
        SELECT CHR(code) AS entity_group
        FROM UNNEST(GENERATE_ARRAY(65, 90)) AS code
    ),
    entity_ids AS (
        SELECT FORMAT('%06d', id) AS entity_id
        FROM UNNEST(GENERATE_ARRAY(1, 5000)) AS id
    ),
    entities AS (
        SELECT entity_group, entity_id
        FROM entity_groups CROSS JOIN entity_ids
    ),
    raw_features AS (
        SELECT
            entity_group,
            entity_id,
            CAST(FLOOR(RAND() * 1000) AS INT64) AS feature_int_1,
            CAST(FLOOR(RAND() * 10000) AS INT64) AS feature_int_2,
            CAST(FLOOR(RAND() * 100) AS INT64) AS feature_int_3,
            CAST(FLOOR(RAND() * 1000000) AS INT64) AS feature_int_4,
            CAST(MOD(ABS(FARM_FINGERPRINT(CONCAT(entity_group, entity_id))), 500) AS INT64) AS feature_int_5,
            ROUND(RAND() * 100, 4) AS feature_float_1,
            ROUND(RAND() * 1000, 4) AS feature_float_2,
            IF(RAND() > 0.05, ROUND(RAND() * 50 - 25, 4), NULL) AS feature_float_3,
            IF(RAND() > 0.05, ROUND(RAND() * 10, 4), NULL) AS feature_float_4,
            ROUND(RAND(), 6) AS feature_float_5,
            CAST(ROUND(RAND() * 99999, 2) AS NUMERIC) AS feature_numeric_1,
            IF(RAND() > 0.05, CAST(ROUND(RAND() * 9999999, 4) AS NUMERIC), NULL) AS feature_numeric_2,
            CAST(ROUND(RAND() * 100, 6) AS NUMERIC) AS feature_numeric_3,
            CAST(ROUND(RAND() * 999999999999, 2) AS BIGNUMERIC) AS feature_bignumeric_1,
            CAST(ROUND(RAND() * 99999999, 8) AS BIGNUMERIC) AS feature_bignumeric_2,
            CONCAT('cat_', CAST(CAST(FLOOR(RAND() * 20) AS INT64) AS STRING)) AS feature_string_1,
            IF(RAND() > 0.05, CONCAT('type_', CAST(CAST(FLOOR(RAND() * 50) AS INT64) AS STRING)), NULL) AS feature_string_2,
            ['alpha', 'beta', 'gamma', 'delta', 'epsilon', 'zeta', 'eta', 'theta'][ORDINAL(CAST(CEIL(RAND() * 8) AS INT64))] AS feature_string_3,
            CONCAT(entity_group, '_', ['north', 'south', 'east', 'west'][ORDINAL(CAST(CEIL(RAND() * 4) AS INT64))]) AS feature_string_4,
            FORMAT('%s-%06d', entity_group, CAST(FLOOR(RAND() * 999999) AS INT64)) AS feature_string_5,
            CAST(CONCAT('bytes_', entity_group, entity_id) AS BYTES) AS feature_bytes_1,
            CAST(FORMAT('%06d', CAST(FLOOR(RAND() * 999999) AS INT64)) AS BYTES) AS feature_bytes_2,
            RAND() > 0.5 AS feature_bool_1,
            RAND() > 0.7 AS feature_bool_2,
            MOD(CAST(FLOOR(RAND() * 10) AS INT64), 2) = 0 AS feature_bool_3,
            DATE_ADD(DATE '2020-01-01', INTERVAL CAST(FLOOR(RAND() * 1825) AS INT64) DAY) AS feature_date_1,
            IF(RAND() > 0.05, DATE_ADD(DATE '2022-06-01', INTERVAL CAST(FLOOR(RAND() * 730) AS INT64) DAY), NULL) AS feature_date_2,
            DATE_ADD(DATE '2023-01-01', INTERVAL CAST(FLOOR(RAND() * 365) AS INT64) DAY) AS feature_date_3,
            DATETIME_ADD(DATETIME '2023-01-01 00:00:00', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_datetime_1,
            DATETIME_ADD(DATETIME '2024-01-01 00:00:00', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_datetime_2,
            TIME_ADD(TIME '00:00:00', INTERVAL CAST(FLOOR(RAND() * 86400) AS INT64) SECOND) AS feature_time_1,
            TIME_ADD(TIME '00:00:00', INTERVAL CAST(FLOOR(RAND() * 86400) AS INT64) SECOND) AS feature_time_2,
            TIMESTAMP_ADD(TIMESTAMP '2023-01-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_timestamp_1,
            TIMESTAMP_ADD(TIMESTAMP '2024-01-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_timestamp_2,
            TIMESTAMP_ADD(TIMESTAMP '2024-06-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 262800) AS INT64) MINUTE) AS feature_timestamp_3,
            ST_GEOGPOINT(ROUND(-180 + RAND() * 360, 6), ROUND(-90 + RAND() * 180, 6)) AS feature_geo_1,
            ST_GEOGPOINT(ROUND(-122.5 + RAND() * 5, 6), ROUND(37.0 + RAND() * 3, 6)) AS feature_geo_2,
            JSON_OBJECT('score', ROUND(RAND() * 100, 2), 'label', CONCAT('class_', CAST(CAST(FLOOR(RAND() * 10) AS INT64) AS STRING)), 'active', RAND() > 0.5) AS feature_json_1,
            JSON_OBJECT('x', ROUND(RAND() * 10, 4), 'y', ROUND(RAND() * 10, 4), 'tag', entity_group) AS feature_json_2,
            [CAST(FLOOR(RAND() * 100) AS INT64), CAST(FLOOR(RAND() * 100) AS INT64), CAST(FLOOR(RAND() * 100) AS INT64)] AS feature_array_int_1,
            [CAST(FLOOR(RAND() * 1000) AS INT64), CAST(FLOOR(RAND() * 1000) AS INT64)] AS feature_array_int_2,
            [['red', 'blue', 'green', 'yellow'][ORDINAL(CAST(CEIL(RAND() * 4) AS INT64))], ['circle', 'square', 'triangle'][ORDINAL(CAST(CEIL(RAND() * 3) AS INT64))]] AS feature_array_string_1,
            STRUCT(ROUND(RAND() * 100, 2) AS value, CONCAT('unit_', CAST(CAST(FLOOR(RAND() * 5) AS INT64) AS STRING)) AS unit, RAND() > 0.5 AS is_valid) AS feature_struct_1,
            STRUCT(CAST(FLOOR(RAND() * 1000) AS INT64) AS count, ROUND(RAND(), 4) AS ratio) AS feature_struct_2,
            [STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name), STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name), STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name)] AS feature_array_struct_1,
            RANGE(DATE_ADD(DATE '2020-01-01', INTERVAL CAST(FLOOR(RAND() * 1000) AS INT64) DAY), DATE_ADD(DATE '2023-01-01', INTERVAL CAST(FLOOR(RAND() * 730) AS INT64) DAY)) AS feature_range_date_1,
        {float_cols_sql}
        FROM entities
    )
    SELECT *,
        CASE WHEN (
            0.4 * feature_int_1 / 999.0 +
            0.3 * feature_float_1 / 100.0 +
            0.15 * (COALESCE(feature_float_3, 0.0) + 25.0) / 50.0 +
            0.15 * CAST(feature_bool_1 AS INT64)
            + (RAND() - 0.5) * 0.3
        ) > 0.6 THEN 'positive' ELSE 'negative' END AS label
    FROM raw_features
    """

    bq.query(dense_query).result()
    print('  dense_features created.')

if 'sparse_events' in missing:
    print('Creating sparse_events table (~2.6M rows)...')

    sparse_query = f"""
    CREATE OR REPLACE TABLE `{BQ_PROJECT}.{BQ_DATASET}.sparse_events` AS
    WITH
    entities AS (
        SELECT entity_group, entity_id
        FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
    ),
    events_per_entity AS (
        SELECT entity_group, entity_id, event_num
        FROM entities CROSS JOIN UNNEST(GENERATE_ARRAY(1, 20)) AS event_num
    )
    SELECT
        entity_group, entity_id,
        TIMESTAMP_SUB(CURRENT_TIMESTAMP(), INTERVAL CAST(FLOOR(RAND() * 90 * 24 * 60) AS INT64) MINUTE) AS event_timestamp,
        ['purchase', 'view', 'click', 'search', 'return'][ORDINAL(CAST(CEIL(RAND() * 5) AS INT64))] AS event_type,
        ROUND(RAND() * 100, 2) AS metric_1,
        ROUND(RAND() * 50, 2) AS metric_2,
        ROUND(RAND() * 200, 2) AS metric_3,
        ROUND(RAND() * 10, 2) AS metric_4,
        ROUND(RAND() * 1000, 2) AS metric_5,
        IF(RAND() > 0.9, 'positive', 'negative') AS label
    FROM events_per_entity
    """

    bq.query(sparse_query).result()
    print('  sparse_events created.')

if not missing:
    print('Both tables already exist — no action taken.')
````

## Spanner Instance

### Cloud Spanner Architecture

[Cloud Spanner](https://cloud.google.com/spanner/docs/overview) is a fully managed, globally distributed relational database with strong consistency, SQL support, and automatic horizontal scaling.

```
Project
└── Instance              ← compute capacity (Processing Units) + regional/multi-regional config
    └── Database          ← schema (tables, indexes), access control
        ├── Table 1       ← rows distributed across splits
        ├── Table 2       ← can be interleaved with parent table
        └── Indexes       ← secondary indexes for non-PK lookups
```

| Concept | What It Is | Documentation |
|---------|-----------|---------------|
| **[Instance](https://cloud.google.com/spanner/docs/instances)** | Container for databases. Determines compute capacity (Processing Units) and regional/multi-regional configuration. | [Instances](https://cloud.google.com/spanner/docs/instances) |
| **[Edition](https://cloud.google.com/spanner/docs/editions-overview)** | Feature tier: STANDARD, ENTERPRISE, or ENTERPRISE_PLUS. **ENTERPRISE+ is required for vector (ScaNN) indexes and ANN search** used in the Vector Embeddings and Production notebooks. | [Editions](https://cloud.google.com/spanner/docs/editions-overview) |
| **[Processing Units](https://cloud.google.com/spanner/docs/compute-capacity)** | Measure of compute capacity. 1000 PU = 1 node. Minimum 100 PU, increments of 100. More PUs = higher read/write throughput. | [Compute capacity](https://cloud.google.com/spanner/docs/compute-capacity) |
| **[Database](https://cloud.google.com/spanner/docs/getting-started/set-up)** | Holds tables, indexes, and views. Schema defined via DDL. Each database has its own schema and access control. | [Schema and data model](https://cloud.google.com/spanner/docs/schema-and-data-model) |
| **[Primary Key](https://cloud.google.com/spanner/docs/schema-and-data-model#primary_keys)** | Composite primary keys determine row ordering and split distribution. Rows with similar key prefixes are stored together. | [Primary keys](https://cloud.google.com/spanner/docs/schema-and-data-model#primary_keys) |
| **[Interleaved Tables](https://cloud.google.com/spanner/docs/schema-and-data-model#creating-interleaved-tables)** | Parent-child tables stored together on the same split for fast JOINs. | [Interleaved tables](https://cloud.google.com/spanner/docs/schema-and-data-model#creating-interleaved-tables) |
| **[Secondary Indexes](https://cloud.google.com/spanner/docs/secondary-indexes)** | Enable efficient lookups on non-primary-key columns. Unlike Bigtable, Spanner supports arbitrary indexed queries. | [Secondary indexes](https://cloud.google.com/spanner/docs/secondary-indexes) |

**Key difference from Bigtable:** Spanner is a relational database with SQL, strongly typed schemas, ACID transactions, and secondary indexes. Bigtable is a NoSQL wide-column store with a single row key index. For feature stores, this means Spanner can query by any indexed column, JOIN feature tables at serving time, and update multiple entities atomically.

### Create Database and Feature Table

This database uses the **GoogleSQL dialect** (Spanner's default; the alternative is the PostgreSQL dialect). The dialect choice is load-bearing later: BigQuery's continuous-query reverse-ETL to Spanner is **not supported for PostgreSQL-dialect** databases (see the Synchronization notebook).

**Composite primary key & splits.** The key `(entity_group, entity_id)` is first-class — no row-key concatenation. Spanner shards the table into *splits* along this key, and the `entity_group` prefix co-locates an entire group (A–Z) onto contiguous splits. A monotonically increasing leading key would hotspot a single split; our 26-way group prefix spreads load across splits.


````python
# Build DDL for the feature table
# Map the most important feature columns to Spanner types

float_cols_ddl = ',\n'.join(
    f'    feature_float_{i} FLOAT64' for i in range(6, 181)
)

DDL = f"""CREATE TABLE {SPANNER_TABLE} (
    entity_group STRING(1) NOT NULL,
    entity_id STRING(6) NOT NULL,

    -- INT64 features
    feature_int_1 INT64,
    feature_int_2 INT64,
    feature_int_3 INT64,
    feature_int_4 INT64,
    feature_int_5 INT64,

    -- FLOAT64 features (first 5)
    feature_float_1 FLOAT64,
    feature_float_2 FLOAT64,
    feature_float_3 FLOAT64,
    feature_float_4 FLOAT64,
    feature_float_5 FLOAT64,

    -- NUMERIC features
    feature_numeric_1 NUMERIC,
    feature_numeric_2 NUMERIC,
    feature_numeric_3 NUMERIC,

    -- STRING features
    feature_string_1 STRING(MAX),
    feature_string_2 STRING(MAX),
    feature_string_3 STRING(MAX),
    feature_string_4 STRING(MAX),
    feature_string_5 STRING(MAX),

    -- BYTES features
    feature_bytes_1 BYTES(MAX),
    feature_bytes_2 BYTES(MAX),

    -- BOOL features
    feature_bool_1 BOOL,
    feature_bool_2 BOOL,
    feature_bool_3 BOOL,

    -- DATE features
    feature_date_1 DATE,
    feature_date_2 DATE,
    feature_date_3 DATE,

    -- TIMESTAMP features (BQ DATETIME has no zone; interpreted as UTC on load)
    feature_datetime_1 TIMESTAMP,
    feature_datetime_2 TIMESTAMP,
    feature_timestamp_1 TIMESTAMP,
    feature_timestamp_2 TIMESTAMP,
    feature_timestamp_3 TIMESTAMP,

    -- TIME features (stored as STRING)
    feature_time_1 STRING(15),
    feature_time_2 STRING(15),

    -- GEOGRAPHY features (stored as WKT STRING)
    feature_geo_1 STRING(MAX),
    feature_geo_2 STRING(MAX),

    -- JSON features
    feature_json_1 JSON,
    feature_json_2 JSON,

    -- ARRAY features
    feature_array_int_1 ARRAY<INT64>,
    feature_array_int_2 ARRAY<INT64>,
    feature_array_string_1 ARRAY<STRING(MAX)>,

    -- STRUCT and complex types (stored as JSON)
    feature_struct_1 JSON,
    feature_struct_2 JSON,
    feature_array_struct_1 JSON,
    feature_range_date_1 JSON,

    -- Bulk FLOAT64 features
{float_cols_ddl},

    -- Label
    label STRING(10),

    -- Metadata
    updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true),
) PRIMARY KEY (entity_group, entity_id)"""

print(f'DDL: {len(DDL):,} characters')
print(f'Table: {SPANNER_TABLE}')
print(f'Primary key: (entity_group, entity_id)')
````

````python
database = instance.database(SPANNER_DATABASE, ddl_statements=[DDL])

if not database.exists():
    operation = database.create()
    operation.result()  # Wait for creation
    print(f'Created database: {SPANNER_DATABASE}')
    print(f'  Table: {SPANNER_TABLE} (composite PK: entity_group, entity_id)')
else:
    print(f'Database exists: {SPANNER_DATABASE}')
    # Check if table exists via INFORMATION_SCHEMA
    with database.snapshot() as snapshot:
        results = snapshot.execute_sql(
            "SELECT TABLE_NAME FROM INFORMATION_SCHEMA.TABLES WHERE TABLE_SCHEMA = ''"
        )
        tables = [row[0] for row in results]
        print(f'  Tables: {tables}')
````

### Verify Spanner Schema

Query `INFORMATION_SCHEMA` to inspect the table schema. Unlike Bigtable (where you store a `#schema` metadata row), Spanner's schema is self-describing via `INFORMATION_SCHEMA`.

````python
with database.snapshot() as snapshot:
    results = snapshot.execute_sql("""
        SELECT COLUMN_NAME, SPANNER_TYPE, IS_NULLABLE, ORDINAL_POSITION
        FROM INFORMATION_SCHEMA.COLUMNS
        WHERE TABLE_NAME = @table
        ORDER BY ORDINAL_POSITION
    """, params={'table': SPANNER_TABLE}, param_types={'table': spanner.param_types.STRING})

    columns = list(results)

print(f'Table {SPANNER_TABLE} schema ({len(columns)} columns):\n')
print('First 30 columns:')
for col in columns[:30]:
    nullable = '' if col[2] == 'YES' else ' NOT NULL'
    print(f'  {col[3]:>4}  {col[0]:<35s} {col[1]:<25s}{nullable}')

print(f'\n  ... ({len(columns) - 40} more columns) ...\n')
print('Last 10 columns:')
for col in columns[-10:]:
    nullable = '' if col[2] == 'YES' else ' NOT NULL'
    print(f'  {col[3]:>4}  {col[0]:<35s} {col[1]:<25s}{nullable}')

# Type distribution
type_counts = {}
for col in columns:
    t = col[1]
    type_counts[t] = type_counts.get(t, 0) + 1

print(f'\nSpanner type distribution:')
for t, count in sorted(type_counts.items(), key=lambda x: -x[1]):
    print(f'  {t:<30s} {count:>3}')
````

Load a sample of the BigQuery data into Spanner to verify the schema and test reads. The full sync is covered in Notebook 1 (Fundamentals) and Notebook 2 (Synchronization).

````python
import time

# Check if data already exists
with database.snapshot() as snapshot:
    result = list(snapshot.execute_sql(
        f'SELECT COUNT(*) FROM {SPANNER_TABLE}'
    ))
    existing_count = result[0][0]

if existing_count > 0:
    print(f'Table already has {existing_count:,} rows — skipping initial load.')
else:
    print('Loading entity group A (5,000 rows) as a sample...')
    start = time.perf_counter()

    # Read sample from BigQuery
    sample_df = bq.query(f'''
        SELECT
            entity_group, entity_id,
            feature_int_1, feature_int_2, feature_int_3, feature_int_4, feature_int_5,
            feature_float_1, feature_float_2, feature_float_3, feature_float_4, feature_float_5,
            feature_numeric_1, feature_numeric_2, feature_numeric_3,
            feature_string_1, feature_string_2, feature_string_3, feature_string_4, feature_string_5,
            feature_bool_1, feature_bool_2, feature_bool_3,
            feature_date_1, feature_date_2, feature_date_3,
            label
        FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
        WHERE entity_group = 'A'
        ORDER BY entity_id
    ''').to_dataframe()

    # Prepare columns for Spanner insert (subset of all columns)
    insert_columns = [
        'entity_group', 'entity_id',
        'feature_int_1', 'feature_int_2', 'feature_int_3', 'feature_int_4', 'feature_int_5',
        'feature_float_1', 'feature_float_2', 'feature_float_3', 'feature_float_4', 'feature_float_5',
        'feature_numeric_1', 'feature_numeric_2', 'feature_numeric_3',
        'feature_string_1', 'feature_string_2', 'feature_string_3', 'feature_string_4', 'feature_string_5',
        'feature_bool_1', 'feature_bool_2', 'feature_bool_3',
        'feature_date_1', 'feature_date_2', 'feature_date_3',
        'label',
        'updated_at',
    ]

    # Convert DataFrame rows to Spanner-compatible values
    from decimal import Decimal
    import datetime

    def to_spanner_row(row):
        values = []
        for col in insert_columns[:-1]:  # Skip updated_at
            val = row[col]
            if pd.isna(val):
                values.append(None)
            elif isinstance(val, (np.integer,)):
                values.append(int(val))
            elif isinstance(val, (np.floating,)):
                values.append(float(val))
            elif isinstance(val, (np.bool_,)):
                values.append(bool(val))
            elif isinstance(val, Decimal):
                values.append(val)
            elif isinstance(val, datetime.date) and not isinstance(val, datetime.datetime):
                values.append(val)
            else:
                values.append(val)
        values.append(spanner.COMMIT_TIMESTAMP)  # updated_at
        return values

    rows = [to_spanner_row(row) for _, row in sample_df.iterrows()]

    # Insert in batches of 500 (Spanner mutation limit is 20,000 per commit)
    batch_size = 500
    for i in range(0, len(rows), batch_size):
        batch = rows[i:i + batch_size]
        with database.batch() as batcher:
            batcher.insert(
                table=SPANNER_TABLE,
                columns=insert_columns,
                values=batch,
            )

    elapsed = time.perf_counter() - start
    print(f'  Loaded {len(rows):,} rows in {elapsed:.1f}s')
    print(f'  Batch size: {batch_size}, batches: {(len(rows) + batch_size - 1) // batch_size}')
````

### Quick Latency Test

A first point-read latency baseline. All measurements in this series are taken at **100 PU, regional (`regional-us-central1`), strong reads**, from a client in the same region. Latency depends heavily on these conditions — we restate them wherever we benchmark.


````python
import time

# Point read latency — strong reads, 100 PU, regional, same-region client.
# 50 samples so the percentiles mean something (10 is too few for a median).
def percentiles(xs):
    s = sorted(xs); n = len(s)
    p = lambda q: s[min(n - 1, int(q * n))]
    return p(0.50), p(0.95), p(0.99)

latencies = []
for i in range(50):
    entity_id = f'{(i % 5000) + 1:06d}'
    start = time.perf_counter()
    with database.snapshot() as snapshot:
        result = list(snapshot.execute_sql(
            f"SELECT * FROM {SPANNER_TABLE} WHERE entity_group = 'A' AND entity_id = @eid",
            params={'eid': entity_id},
            param_types={'eid': spanner.param_types.STRING},
        ))
    latencies.append((time.perf_counter() - start) * 1000)

p50, p95, p99 = percentiles(latencies)
print('Point read latency (50 reads, strong consistency, 100 PU, regional):')
print(f'  p50: {p50:>7.1f} ms')
print(f'  p95: {p95:>7.1f} ms')
print(f'  p99: {p99:>7.1f} ms')
print(f'  min: {min(latencies):>7.1f} ms   max: {max(latencies):>7.1f} ms')
````

## Ways to Access Spanner

Spanner offers multiple interfaces for reading and writing data. The SQL interface is its primary advantage over key-value stores like Bigtable.

### Client Libraries

| Language | Client Library | Best For |
|----------|---------------|----------|
| **Python** | [`google-cloud-spanner`](https://cloud.google.com/python/docs/reference/spanner/latest) | Notebooks, data science, serving — **used throughout this series** |

| **Go** | [`cloud.google.com/go/spanner`](https://cloud.google.com/go/docs/reference/cloud.google.com/go/spanner/latest) | Microservices, CLI tools |
| **Node.js** | [`@google-cloud/spanner`](https://cloud.google.com/nodejs/docs/reference/spanner/latest) | Web backends, serverless functions |

### SQL Interfaces

| Interface | Description |
|-----------|------------|
| **[Spanner Studio](https://cloud.google.com/spanner/docs/spanner-studio)** | Visual SQL editor in the Cloud Console |
| **[gcloud spanner](https://cloud.google.com/sdk/gcloud/reference/spanner)** | CLI for admin operations and SQL queries |
| **[INFORMATION_SCHEMA](https://cloud.google.com/spanner/docs/information-schema)** | Self-describing schema — query table structure, indexes, constraints |

### Key Difference from Bigtable

| Capability | Spanner | Bigtable |
|---|---|---|
| Query language | Full SQL (JOINs, WHERE, GROUP BY) | Row key lookups + limited GoogleSQL |
| Schema | Strongly typed, DDL-defined | Schema-free (any column qualifier) |
| Indexes | Secondary indexes on any column | Row key only |
| Transactions | Multi-row, multi-table ACID | Single-row atomic |
| Admin | SQL DDL (`ALTER TABLE`) | Admin API (`create_column_family`) |

### Routing & priority (Spanner's app-profile analog)

Bigtable uses **app profiles** to route requests and set priority. Spanner's equivalents are per-request, not a separate object:

| Bigtable app-profile concept | Spanner equivalent |
|---|---|
| Routing policy (single vs multi-cluster) | **Directed reads** (`DirectedReadOptions`) — pin reads to a replica/region (see Multi-Region NB) |
| Priority (Data Boost / standard) | **Request priority** (`RequestOptions` `PRIORITY_LOW/MEDIUM/HIGH`) |
| Isolation of workloads | Separate databases or priority tiers |
