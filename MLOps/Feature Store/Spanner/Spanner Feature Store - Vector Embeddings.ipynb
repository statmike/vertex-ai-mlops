# Spanner Feature Store — Vector Embeddings

Spanner has **native vector search**. You do *not* need to pull vectors over the wire and run cosine similarity in numpy. Spanner ships two server-side approaches in GoogleSQL:

1. **Exact KNN** via the scalar distance functions `COSINE_DISTANCE()`, `EUCLIDEAN_DISTANCE()`, and `DOT_PRODUCT()` — used as `ORDER BY COSINE_DISTANCE(col, @q) LIMIT k`, and freely combined with a `WHERE` clause.
2. **Approximate nearest neighbor (ANN)** via `APPROX_COSINE_DISTANCE()` / `APPROX_EUCLIDEAN_DISTANCE()` / `APPROX_DOT_PRODUCT()`, backed by a **ScaNN `VECTOR INDEX`** — sub-linear search over large corpora with a tunable recall/latency tradeoff.

**Spanner's unique angle vs. a dedicated vector DB:** combine a SQL **metadata filter** and a **vector distance** in *one* query — `WHERE category='electronics' AND score > 50 ORDER BY COSINE_DISTANCE(embedding, @q) LIMIT 10` — with full ACID consistency and no client round-trip. A dedicated vector store can only post-filter or pre-tag; Spanner runs the relational predicate and the KNN together in the same transaction-consistent statement.

Client-side numpy similarity is shown **once, as a contrast baseline** near the end — it is slower and moves every vector across the network. Prefer the native SQL paths.

**What you'll learn:**
- Store embeddings as `ARRAY<FLOAT32>(vector_length=>N) NOT NULL`
- Native **exact KNN** with `COSINE_DISTANCE` / `EUCLIDEAN_DISTANCE` / `DOT_PRODUCT` in `ORDER BY ... LIMIT k`
- **SQL pre-filtering + KNN in one statement** — Spanner's differentiator
- **ANN** with a ScaNN `VECTOR INDEX` and `APPROX_*` functions, sweeping `num_leaves_to_search` and measuring **recall vs. latency**
- How the native path compares to Bigtable, Valkey, and client-side numpy

**Sources (verified):**
- KNN scalar functions: https://cloud.google.com/spanner/docs/find-k-nearest-neighbors
- ANN / `APPROX_*` functions: https://cloud.google.com/spanner/docs/find-approximate-nearest-neighbors
- `CREATE VECTOR INDEX` (ScaNN): https://cloud.google.com/spanner/docs/vector-indexes

**Prerequisites:** Run [Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) and [Fundamentals](./Spanner%20Feature%20Store%20-%20Fundamentals.ipynb) first. ANN (`APPROX_*` + `VECTOR INDEX`) requires the Spanner **Enterprise** or **Enterprise Plus** edition.

## 1. Create the Embedding Table (FLOAT32 + vector_length)

The embedding column must be `ARRAY<FLOAT32>(vector_length=>N) NOT NULL` to be eligible for a ScaNN `VECTOR INDEX`. We use **64-dim** vectors and add metadata columns (`category`, `score`) so we can demonstrate Spanner's combined *filter + KNN* query. Two secondary indexes that **`STORING` the embedding** let `WHERE category=...` / `WHERE score>...` KNN queries run as index-only scans (per the KNN docs' secondary-index pattern).

````python
embedding_ddl = [
    """CREATE TABLE entity_embeddings (
        entity_group STRING(1) NOT NULL,
        entity_id    STRING(6) NOT NULL,
        embedding    ARRAY<FLOAT32>(vector_length=>64) NOT NULL,
        category     STRING(20),
        score        FLOAT64,
        updated_at   TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
    ) PRIMARY KEY (entity_group, entity_id)""",
    # STORING(embedding) -> index-only scan for filtered exact-KNN queries
    "CREATE INDEX idx_emb_category ON entity_embeddings(category) STORING (embedding)",
    "CREATE INDEX idx_emb_score    ON entity_embeddings(score)    STORING (embedding)",
]

try:
    database.update_ddl(embedding_ddl).result()
    print('Created entity_embeddings (ARRAY<FLOAT32>(vector_length=>64)) + category/score indexes.')
except Exception as e:
    msg = str(e)
    if 'Duplicate' in msg or 'already exists' in msg:
        print('entity_embeddings already exists — reusing it.')
    else:
        raise
````

---
## 2. Load Clustered Embeddings

Pure random vectors are nearly equidistant in high dimensions (curse of dimensionality), which makes "nearest neighbor" almost arbitrary and **recall** meaningless. Real embeddings *cluster*, so we model that: each vector = one of ~50 cluster centroids + small noise. We **L2-normalize** and cast to **float32**, then write with `batch.insert_or_update`.

We load groups **A–B (~10K rows)** — enough for meaningful KNN while keeping the notebook fast. `category` gets one of a few values and `score` a uniform 0–100.

````python
EMBEDDING_DIM = 64

with database.snapshot() as snap:
    emb_count = list(snap.execute_sql('SELECT COUNT(*) FROM entity_embeddings'))[0][0]

if emb_count > 0:
    print(f'Embeddings already loaded: {emb_count:,} rows — skipping load.')
    SKIP_EMB_LOAD = True
else:
    SKIP_EMB_LOAD = False
    print('Loading clustered embeddings...')
````

````python
if not SKIP_EMB_LOAD:
    np.random.seed(42)
    categories = ['electronics', 'clothing', 'food', 'sports', 'books']

    # Cluster structure: ~50 centroids + small noise -> tight, separable clusters,
    # so KNN and ANN recall are actually meaningful.
    N_CLUSTERS = 50
    centroids = np.random.randn(N_CLUSTERS, EMBEDDING_DIM).astype(np.float32)
    centroids /= np.linalg.norm(centroids, axis=1, keepdims=True)
    NOISE = 0.15

    batch_size = 500
    total = 0
    for group_code in range(65, 67):  # groups A, B  (~10K rows)
        group = chr(group_code)
        values = []
        for i in range(1, 5001):
            eid = f'{i:06d}'
            cluster = (group_code * 5000 + i) % N_CLUSTERS
            emb = centroids[cluster] + NOISE * np.random.randn(EMBEDDING_DIM).astype(np.float32)
            emb = emb / np.linalg.norm(emb)
            # ARRAY<FLOAT32> param/value: pass plain python floats; Spanner stores as FLOAT32.
            emb = [float(x) for x in emb.astype(np.float32)]
            category = categories[(group_code * 5000 + i) % len(categories)]
            score = float(np.random.uniform(0, 100))
            values.append([group, eid, emb, category, score, spanner.COMMIT_TIMESTAMP])

            if len(values) >= batch_size:
                with database.batch() as batch:
                    batch.insert_or_update(
                        table='entity_embeddings',
                        columns=['entity_group', 'entity_id', 'embedding', 'category', 'score', 'updated_at'],
                        values=values)
                total += len(values); values = []
        if values:
            with database.batch() as batch:
                batch.insert_or_update(
                    table='entity_embeddings',
                    columns=['entity_group', 'entity_id', 'embedding', 'category', 'score', 'updated_at'],
                    values=values)
            total += len(values)
        print(f'  Group {group}: loaded ({total:,} total)')
    print(f'\nLoad complete: {total:,} rows, {EMBEDDING_DIM}-dim float32 vectors')
````

---
## 3. Native Exact KNN — `ORDER BY COSINE_DISTANCE(...) LIMIT k`

This is the server-side replacement for client-side numpy. Spanner computes the distance for every candidate and returns the top *k* — **the vectors never leave the database**. We pass the query vector as an `ARRAY<FLOAT32>` query parameter.

> **Param type note:** GoogleSQL distance functions expect `ARRAY<FLOAT32>` to match the column. The Python client exposes the FLOAT32 array param type as `spanner.param_types.Array(spanner.param_types.FLOAT32)`. The cell below probes for it and falls back to `FLOAT64` if the installed client version lacks the FLOAT32 scalar type.

````python
# Pick a real stored vector as the query (so it lands near actual data).
with database.snapshot() as snap:
    qrow = list(snap.execute_sql(
        "SELECT embedding FROM entity_embeddings WHERE entity_group='A' AND entity_id='000001'"))
query_vec = [float(x) for x in qrow[0][0]]
print(f'Query vector: dim={len(query_vec)} (entity A/000001), first 5 = {[round(x,3) for x in query_vec[:5]]}')

# Resolve the ARRAY<FLOAT32> param type, with a FLOAT64 fallback for older clients.
pt = spanner.param_types
if hasattr(pt, 'FLOAT32'):
    Q_ARRAY_TYPE = pt.Array(pt.FLOAT32)
    print('Using ARRAY<FLOAT32> query param type.')
else:
    Q_ARRAY_TYPE = pt.Array(pt.FLOAT64)
    print('FLOAT32 scalar param type not in this client version — using ARRAY<FLOAT64> param (Spanner coerces).')
````

````python
def knn_exact(distance_fn, k=10, where='', extra_params=None, extra_types=None):
    sql = f"""
        SELECT entity_group, entity_id, {distance_fn}(embedding, @q) AS dist
        FROM entity_embeddings
        {where}
        ORDER BY dist
        LIMIT {k}
    """
    params = {'q': query_vec}
    types  = {'q': Q_ARRAY_TYPE}
    if extra_params: params.update(extra_params)
    if extra_types:  types.update(extra_types)
    start = time.perf_counter()
    with database.snapshot() as snap:
        rows = list(snap.execute_sql(sql, params=params, param_types=types))
    ms = (time.perf_counter() - start) * 1000
    return rows, ms

rows, ms = knn_exact('COSINE_DISTANCE', k=10)
print(f'Native exact KNN — COSINE_DISTANCE (server-side, {len(rows)} results in {ms:.0f} ms):\n')
for g, e, d in rows:
    print(f'  {g}/{e}: cosine_distance = {d:.4f}')

# Keep the exact COSINE top-10 keys as ground truth for the ANN recall test later.
exact_cosine_keys = [(g, e) for g, e, _ in rows]
````

````python
# The other two distance functions use the identical pattern.
for fn in ('EUCLIDEAN_DISTANCE', 'DOT_PRODUCT'):
    rws, t = knn_exact(fn, k=5)
    print(f'{fn} (top 5, {t:.0f} ms):')
    for g, e, d in rws:
        print(f'  {g}/{e}: {fn.lower()} = {d:.4f}')
    print()
print('Note: DOT_PRODUCT is a valid *distance* ranking only when all embeddings are L2-normalized (they are here).')
````

---
## 4. SQL Pre-Filtering + KNN in ONE Query — Spanner's Unique Angle

Here is the differentiator. The relational predicate (`category`, `score`) and the vector ranking run **in the same statement**, transaction-consistent, with **no client round-trip and no second system**:

```sql
SELECT entity_group, entity_id, COSINE_DISTANCE(embedding, @q) AS dist
FROM entity_embeddings
WHERE category = 'electronics' AND score > 50
ORDER BY dist
LIMIT 10
```

A dedicated vector store would have to either over-fetch then filter client-side, or maintain separate tag indexes. Spanner just adds a `WHERE`.

````python
rows, ms = knn_exact(
    'COSINE_DISTANCE', k=10,
    where="WHERE category = @cat AND score > @minscore",
    extra_params={'cat': 'electronics', 'minscore': 50.0},
    extra_types={'cat': spanner.param_types.STRING, 'minscore': spanner.param_types.FLOAT64})

print(f'Filtered KNN — category=electronics AND score>50, in ONE statement ({ms:.0f} ms):\n')
for g, e, d in rows:
    print(f'  {g}/{e}: cosine_distance = {d:.4f}')
````

````python
# Filter-size latency sweep: no filter vs category vs category+score.
sweep = [
    ('No filter',          '',                                                 None, None),
    ('category only',      'WHERE category = @cat',                            {'cat': 'electronics'},
                            {'cat': spanner.param_types.STRING}),
    ('category + score',   'WHERE category = @cat AND score > @minscore',
                            {'cat': 'electronics', 'minscore': 50.0},
                            {'cat': spanner.param_types.STRING, 'minscore': spanner.param_types.FLOAT64}),
]

print(f'{"Filter":.<22s} {"Median latency":>16s}')
sweep_results = []
for desc, where, ep, et in sweep:
    lats = []
    for _ in range(3):
        _, t = knn_exact('COSINE_DISTANCE', k=10, where=where, extra_params=ep, extra_types=et)
        lats.append(t)
    med = sorted(lats)[1]
    sweep_results.append((desc, med))
    print(f'{desc:.<22s} {med:>14.0f}ms')

print('\nThe filter runs server-side before/with the distance computation — narrower filters touch fewer rows.')
````

---
## 5. Approximate Nearest Neighbor (ANN) with a ScaNN `VECTOR INDEX`

For large corpora, exact KNN scans every row. A **`VECTOR INDEX`** (Google's ScaNN, tree-based partitioning) lets `APPROX_*` functions search only a few partitions. DDL:

```sql
CREATE VECTOR INDEX idx_emb_vector
  ON entity_embeddings(embedding)
  OPTIONS (distance_type='COSINE', tree_depth=2, num_leaves=1000)
```

Because `embedding` is declared `NOT NULL`, **no `IS NOT NULL` filter is required**. ANN query rules: the `APPROX_*` distance must be the **sole `ORDER BY` key** and a **`LIMIT`** is mandatory.

> ANN requires Spanner **Enterprise / Enterprise Plus** edition. The index build is processed as a background schema-update operation; `update_ddl(...).result()` returns when the operation completes. We tune `num_leaves` modestly for a ~10K-row demo (production scales `num_leaves` to roughly sqrt(row count)).

````python
ANN_AVAILABLE = True
try:
    print('Creating ScaNN VECTOR INDEX (background schema op; .result() waits for completion)...')
    start = time.perf_counter()
    database.update_ddl([
        """CREATE VECTOR INDEX idx_emb_vector
               ON entity_embeddings(embedding)
               OPTIONS (distance_type='COSINE', tree_depth=2, num_leaves=200)"""
    ]).result()
    print(f'  Vector index ready ({(time.perf_counter()-start):.0f}s).')
except Exception as e:
    msg = str(e)
    if 'Duplicate' in msg or 'already exists' in msg:
        print('  idx_emb_vector already exists.')
    elif 'Enterprise' in msg or 'edition' in msg.lower() or 'not supported' in msg.lower():
        ANN_AVAILABLE = False
        print(f'  ANN not available on this instance edition — skipping ANN section.\n  ({msg[:160]})')
    else:
        raise
````

````python
def knn_approx(num_leaves_to_search, k=10):
    sql = f"""
        SELECT entity_group, entity_id,
               APPROX_COSINE_DISTANCE(embedding, @q,
                   options => JSON '{{"num_leaves_to_search": {num_leaves_to_search}}}') AS dist
        FROM entity_embeddings
        ORDER BY dist
        LIMIT {k}
    """
    start = time.perf_counter()
    with database.snapshot() as snap:
        rows = list(snap.execute_sql(
            sql, params={'q': query_vec}, param_types={'q': Q_ARRAY_TYPE}))
    ms = (time.perf_counter() - start) * 1000
    return rows, ms

if ANN_AVAILABLE:
    rows, ms = knn_approx(num_leaves_to_search=50, k=10)
    print(f'ANN — APPROX_COSINE_DISTANCE, num_leaves_to_search=50 ({ms:.0f} ms):\n')
    for g, e, d in rows:
        print(f'  {g}/{e}: approx_cosine_distance = {d:.4f}')
else:
    print('Skipped: ANN not available on this instance edition.')
````

````python
# Sweep num_leaves_to_search and compute recall vs. the exact COSINE top-10 (section 3).
if ANN_AVAILABLE:
    exact_set = set(exact_cosine_keys)
    K = len(exact_set)
    print(f'{"num_leaves_to_search":.<24s} {"latency":>10s} {"recall@%d" % K:>10s}')
    ann_sweep = []
    for nls in (10, 50, 200):
        lats = []
        approx_keys = None
        for _ in range(3):
            rows, t = knn_approx(num_leaves_to_search=nls, k=K)
            lats.append(t)
            approx_keys = [(g, e) for g, e, _ in rows]
        med = sorted(lats)[1]
        recall = len(exact_set & set(approx_keys)) / len(exact_set)
        ann_sweep.append((nls, med, recall))
        print(f'{nls:.<24d} {med:>8.0f}ms {recall:>9.0%}')
    print('\nHigher num_leaves_to_search searches more partitions: higher recall, higher latency.')
else:
    ann_sweep = []
    print('Skipped: ANN not available on this instance edition.')
````

````python
# Recall-vs-latency tradeoff plot (like the Valkey HNSW sweep).
if ANN_AVAILABLE and ann_sweep:
    fig, ax1 = plt.subplots(figsize=(9, 5))
    nls_vals  = [r[0] for r in ann_sweep]
    lat_vals  = [r[1] for r in ann_sweep]
    rec_vals  = [r[2] * 100 for r in ann_sweep]

    ax1.plot(nls_vals, lat_vals, 'o-', color='#EA4335', label='Latency (ms)')
    ax1.set_xlabel('num_leaves_to_search', fontweight='bold')
    ax1.set_ylabel('Latency (ms)', color='#EA4335', fontweight='bold')
    ax1.tick_params(axis='y', labelcolor='#EA4335')

    ax2 = ax1.twinx()
    ax2.plot(nls_vals, rec_vals, 's--', color='#4285F4', label='Recall (%)')
    ax2.set_ylabel('Recall@K (%)', color='#4285F4', fontweight='bold')
    ax2.set_ylim(0, 105)
    ax2.tick_params(axis='y', labelcolor='#4285F4')

    ax1.set_title('ANN Recall vs. Latency — Spanner ScaNN VECTOR INDEX', fontweight='bold')
    ax1.grid(True, alpha=0.3)
    fig.tight_layout()
    plt.show()
else:
    print('Skipped plot: ANN not available.')
````

## 6. Client-Side numpy — Contrast Baseline ONLY

This is the old "Spanner has no vector search" approach. It pulls **every** vector over the wire and computes `embeddings @ query` in numpy. Shown only so you can see why the native SQL paths above are preferable: **prefer native SQL — this moves all vectors across the network and is slower.**

````python
# BASELINE FOR COMPARISON ONLY — prefer the native SQL KNN/ANN above.
start = time.perf_counter()
with database.snapshot() as snap:
    all_rows = list(snap.execute_sql(
        "SELECT entity_group, entity_id, embedding FROM entity_embeddings"))
read_ms = (time.perf_counter() - start) * 1000

t0 = time.perf_counter()
keys = [(r[0], r[1]) for r in all_rows]
mat  = np.array([r[2] for r in all_rows], dtype=np.float32)   # all vectors over the wire
q    = np.array(query_vec, dtype=np.float32)
sims = mat @ q                                                # cosine (vectors are normalized)
top  = np.argsort(sims)[-10:][::-1]
compute_ms = (time.perf_counter() - t0) * 1000

print(f'Client-side brute force ({len(all_rows):,} vectors pulled to the client):')
print(f'  Read over wire : {read_ms:>7.0f} ms')
print(f'  numpy compute  : {compute_ms:>7.1f} ms')
print(f'  Total          : {read_ms + compute_ms:>7.0f} ms')
print('\nCompare to native COSINE_DISTANCE above — native keeps vectors server-side and returns only top-k.')
````

## 7. Comparison Across Approaches

| Approach | Vector search | Metadata filter | Notes |
|---|---|---|---|
| **Spanner — native exact KNN** | `COSINE/EUCLIDEAN/DOT_PRODUCT` in `ORDER BY ... LIMIT k`, server-side | **SQL `WHERE` on any column, in the *same* query** | ACID; ideal for small/filtered candidate sets |
| **Spanner — native ANN (ScaNN)** | `APPROX_*` + `VECTOR INDEX`, sub-linear | Filter supported alongside ANN (sole `ORDER BY` + `LIMIT`) | Enterprise edition; tune `num_leaves_to_search` for recall/latency |
| **Spanner — SQL pre-filter + KNN** | filter + distance in **one statement** | **The differentiator** — relational predicate + KNN together | No client round-trip, no second system |
| **Bigtable** | `COSINE_DISTANCE()` brute-force (GoogleSQL) | Row-key prefix only | No rich metadata filtering |
| **Valkey** | Native HNSW (`FT.SEARCH`) | TAG / NUMERIC filters | Sub-ms ANN; separate store from your relational data |
| **Client-side numpy** | `embeddings @ q` in the client | Any (but after pulling all vectors) | **Contrast only** — slowest, moves all vectors over the wire |

---
## 8. When to Use Which

- **Exact KNN (`COSINE_DISTANCE` + `ORDER BY ... LIMIT`)** — small or **SQL-filtered** candidate sets. When a `WHERE` narrows to a few thousand rows, exact KNN is simple, precise, and needs no index maintenance. This is the sweet spot for **filter + KNN in one query**.
- **ANN (`APPROX_*` + ScaNN `VECTOR INDEX`)** — **large corpora** where scanning every row is too slow. Accept slightly <100% recall for sub-linear latency; tune `num_leaves_to_search` (and `num_leaves` at build time) for your recall/latency target.
- **Client-side numpy** — **never as the primary path.** Only as a teaching baseline or a one-off on a tiny set already in memory. It defeats the point of a server-side vector engine.
