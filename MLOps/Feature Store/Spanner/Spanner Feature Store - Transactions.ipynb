# Spanner Feature Store — Transactions and Atomic Updates

ACID transactions across rows and tables. This is where Spanner fundamentally differs from Bigtable — Bigtable offers single-row atomicity only, while Spanner provides multi-row, multi-table ACID transactions with configurable consistency.

**What you'll learn:**
- Read-write transactions: update multiple entities atomically
- Optimistic concurrency and automatic abort/retry: no lost updates under contention
- Read-only transactions with `STRONG` consistency vs behavioral stale reads (`EXACT_STALENESS`, `read_timestamp`)
- Commit timestamps: server-assigned, monotonic ordering for feature freshness and time-travel
- Batch DML (`batch_update`), the 80,000-mutation commit limit, and how to count mutations
- Partitioned DML for large-scale, limit-free feature backfills
- Cross-table transactions: update parent and child tables atomically
- Transaction latency, and how lock contention inflates tail latency (p95/p99)

**Feature store use case:** Atomic feature updates during serving — update a user's click count and an item's popularity score together, ensuring consistency even under concurrent updates.

**Key difference from Bigtable:** Bigtable's `ReadModifyWriteRow` and `CheckAndMutateRow` are single-row atomic only. Spanner transactions span rows and tables.

**Prerequisites:** Run [Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) and [Fundamentals](./Spanner%20Feature%20Store%20-%20Fundamentals.ipynb) first.

## 1. Read-Write Transactions

A read-write transaction reads data, performs computation, and writes results atomically. If another transaction modifies the same rows concurrently, Spanner detects the conflict and retries automatically.

### Atomic Feature Update

Read a feature value, compute a new value, and write it back — all atomically. This is the pattern for real-time feature updates (e.g., incrementing a click counter based on the current value).

````python
# Read current value
with database.snapshot() as snapshot:
    before = list(snapshot.execute_sql(f"""
        SELECT feature_int_1, feature_float_1
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id = '000001'
    """))

print(f'Before transaction:')
print(f'  feature_int_1:   {before[0][0]}')
print(f'  feature_float_1: {before[0][1]:.4f}')
````

````python
def atomic_increment(transaction):
    """Read feature_int_1, increment by 1, write back."""
    result = list(transaction.execute_sql(f"""
        SELECT feature_int_1
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id = '000001'
    """))
    current_value = result[0][0]
    new_value = current_value + 1

    transaction.update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
        values=[['A', '000001', new_value, spanner.COMMIT_TIMESTAMP]],
    )
    return new_value

start = time.perf_counter()
new_value = database.run_in_transaction(atomic_increment)
txn_ms = (time.perf_counter() - start) * 1000

print(f'Atomic increment transaction: {txn_ms:.1f} ms')
print(f'  feature_int_1: {before[0][0]} → {new_value}')
````

### Multi-Row Atomic Update

Update two entities in the same transaction — either both succeed or neither does.

````python
def transfer_value(transaction):
    """Atomically transfer feature_float_1 value between two entities."""
    results = list(transaction.execute_sql(f"""
        SELECT entity_id, feature_float_1
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id IN ('000001', '000002')
        ORDER BY entity_id
    """))

    val_1 = results[0][1]
    val_2 = results[1][1]

    transfer_amount = 10.0

    transaction.update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
        values=[
            ['A', '000001', val_1 - transfer_amount, spanner.COMMIT_TIMESTAMP],
            ['A', '000002', val_2 + transfer_amount, spanner.COMMIT_TIMESTAMP],
        ],
    )
    return val_1, val_2, transfer_amount

start = time.perf_counter()
val_1, val_2, amount = database.run_in_transaction(transfer_value)
multi_row_ms = (time.perf_counter() - start) * 1000

print(f'Multi-row atomic transfer: {multi_row_ms:.1f} ms')
print(f'  Entity 000001: {val_1:.2f} → {val_1 - amount:.2f}')
print(f'  Entity 000002: {val_2:.2f} → {val_2 + amount:.2f}')
print(f'  Transfer amount: {amount:.2f}')
print(f'\nIn Bigtable, this would require two separate writes with no atomicity guarantee.')
````

## 1b. Optimistic Concurrency and Automatic Abort/Retry

This is the headline Spanner guarantee. Spanner read-write transactions are **serializable**: concurrent transactions behave as if they ran one at a time. Spanner uses *optimistic concurrency control* — a transaction reads, buffers its writes, and at commit time Spanner verifies that nothing it read was changed by another committed transaction. If it was, the commit **aborts** and `database.run_in_transaction` **automatically retries** the whole function from scratch.

That means a correct read-modify-write loop never loses updates, even under heavy contention — you just pay in retries. Below we prove it: launch many threads that each increment the **same row's** `feature_int_1`, count how many times the transaction body runs (each extra run = a retry after an abort), and confirm the final value increased by *exactly* the number of threads. Then we contrast with a naive non-transactional read-then-write that **does** lose updates.

We operate only on a dedicated reserved scratch entity (`group = 'Z'`, `id = 'TXNT'`) that this section creates and deletes — real feature rows are never touched.

````python
from concurrent.futures import ThreadPoolExecutor
import threading

SCRATCH_GROUP = 'Z'
SCRATCH_ID = 'TXNT'
N_THREADS = 10

def reset_scratch(value=0):
    """Create or reset the reserved scratch row to a known value."""
    with database.batch() as batch:
        batch.insert_or_update(
            table=SPANNER_TABLE,
            columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
            values=[[SCRATCH_GROUP, SCRATCH_ID, value, spanner.COMMIT_TIMESTAMP]],
        )

reset_scratch(0)
print(f'Scratch row {SCRATCH_GROUP}/{SCRATCH_ID} initialized to feature_int_1 = 0')
````

````python
# Thread-safe counter of how many times the transaction BODY executes.
# run_in_transaction re-invokes the function on each abort/retry, so
# (invocations - successes) == number of retried aborts.
invocation_count = {'n': 0}
counter_lock = threading.Lock()

def increment_scratch(transaction):
    with counter_lock:
        invocation_count['n'] += 1
    row = list(transaction.execute_sql(f"""
        SELECT feature_int_1 FROM {SPANNER_TABLE}
        WHERE entity_group = '{SCRATCH_GROUP}' AND entity_id = '{SCRATCH_ID}'
    """))
    current = row[0][0]
    transaction.update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
        values=[[SCRATCH_GROUP, SCRATCH_ID, current + 1, spanner.COMMIT_TIMESTAMP]],
    )

def worker(_):
    database.run_in_transaction(increment_scratch)

start = time.perf_counter()
with ThreadPoolExecutor(max_workers=N_THREADS) as pool:
    list(pool.map(worker, range(N_THREADS)))
elapsed = (time.perf_counter() - start) * 1000

with database.snapshot() as snapshot:
    final = list(snapshot.execute_sql(f"""
        SELECT feature_int_1 FROM {SPANNER_TABLE}
        WHERE entity_group = '{SCRATCH_GROUP}' AND entity_id = '{SCRATCH_ID}'
    """))[0][0]

retries = invocation_count['n'] - N_THREADS
print(f'{N_THREADS} concurrent transactional increments in {elapsed:.0f} ms')
print(f'  Final feature_int_1:      {final}  (expected {N_THREADS})')
print(f'  Transaction invocations:  {invocation_count["n"]}')
print(f'  Aborts / retries:         {retries}')
print()
if final == N_THREADS:
    print(f'No lost updates — serializability held despite {retries} abort(s).')
else:
    print('Unexpected: lost updates under a serializable transaction (should not happen).')
````

### Contrast: naive non-transactional read-then-write loses updates

Now run the *same* concurrent increment, but with no transaction — a plain snapshot read followed by a separate batch write. Each thread reads a value, another thread reads the same value before the first writes, and one of the two increments is silently lost. With this race there is no abort and no retry, so the final value lands **below** the thread count.

````python
reset_scratch(0)

def naive_increment(_):
    # Read in one (snapshot) operation...
    with database.snapshot(multi_use=True) as snapshot:
        current = list(snapshot.execute_sql(f"""
            SELECT feature_int_1 FROM {SPANNER_TABLE}
            WHERE entity_group = '{SCRATCH_GROUP}' AND entity_id = '{SCRATCH_ID}'
        """))[0][0]
    time.sleep(0.01)  # widen the race window
    # ...then write in a separate, unrelated operation. No conflict check.
    with database.batch() as batch:
        batch.update(
            table=SPANNER_TABLE,
            columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
            values=[[SCRATCH_GROUP, SCRATCH_ID, current + 1, spanner.COMMIT_TIMESTAMP]],
        )

with ThreadPoolExecutor(max_workers=N_THREADS) as pool:
    list(pool.map(naive_increment, range(N_THREADS)))

with database.snapshot(multi_use=True) as snapshot:
    naive_final = list(snapshot.execute_sql(f"""
        SELECT feature_int_1 FROM {SPANNER_TABLE}
        WHERE entity_group = '{SCRATCH_GROUP}' AND entity_id = '{SCRATCH_ID}'
    """))[0][0]

print(f'Naive read-then-write final: {naive_final}  (expected {N_THREADS})')
print(f'  Lost updates: {N_THREADS - naive_final}')
print()
print('Lesson: only the transactional version is safe under concurrency.')
print('The same read-modify-write inside run_in_transaction never loses an update;')
print('Spanner aborts and replays the loser instead of overwriting the winner.')
````

## 2. Read-Only Transactions with Staleness

Read-only transactions don't acquire locks, so they're faster and never conflict. Combined with staleness bounds, they can serve from local replicas for even lower latency.

### Staleness Options

| Mode | Description | Use Case |
|---|---|---|
| `strong=True` | Linearizable — sees latest committed data | When you need absolute freshness |
| `exact_staleness=N` | Read as of N seconds ago | Feature serving where slight staleness is OK |
| `max_staleness=N` | Spanner picks freshest available within N | Maximum latency optimization |
| `read_timestamp=T` | Read at a specific timestamp | Reproducible snapshots for training |

````python
query = f"""
    SELECT entity_group, entity_id, feature_float_1, feature_int_1
    FROM {SPANNER_TABLE}
    WHERE entity_group = 'B' AND entity_id = '000100'
"""

n_reads = 20
staleness_configs = [
    ('Strong', {}),
    ('Exact 10s', {'exact_staleness': datetime.timedelta(seconds=10)}),
    ('Exact 30s', {'exact_staleness': datetime.timedelta(seconds=30)}),
    ('Max 15s', {'max_staleness': datetime.timedelta(seconds=15)}),
]

all_latencies = {}
for name, kwargs in staleness_configs:
    latencies = []
    for _ in range(n_reads):
        start = time.perf_counter()
        with database.snapshot(**kwargs) as snapshot:
            list(snapshot.execute_sql(query))
        latencies.append((time.perf_counter() - start) * 1000)
    all_latencies[name] = latencies

print(f'Read-only transaction latency ({n_reads} reads each):\n')
print(f'{"Mode":<15s} {"Min":>8s} {"Median":>8s} {"p95":>8s} {"Max":>8s}')
for name, lats in all_latencies.items():
    s = sorted(lats)
    print(f'{name:<15s} {s[0]:>6.1f}ms {s[n_reads//2]:>6.1f}ms {s[int(n_reads*0.95)]:>6.1f}ms {s[-1]:>6.1f}ms')
````

````python
# Visualize
fig, ax = plt.subplots(figsize=(10, 5))

positions = range(len(all_latencies))
labels = list(all_latencies.keys())
data = [all_latencies[l] for l in labels]
colors = ['#EA4335', '#FBBC04', '#34A853', '#4285F4']

bp = ax.boxplot(data, tick_labels=labels, patch_artist=True)
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('Latency (ms)', fontweight='bold')
ax.set_title('Read Latency by Staleness Mode', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
````

### Staleness is a freshness/latency trade-off, not just latency

The latency table above shows stale reads are *faster*. But the real point of staleness is **what value you see**. Here we prove it behaviorally: write a new value to a scratch row, then immediately read it two ways. A **strong** read is linearizable and sees the new value. An **exact-staleness 15s** read is served as of 15 seconds ago — *before* our write — so it returns the OLD value. Lower latency, bounded staleness, by construction.

````python
STALE_GROUP, STALE_ID = 'Z', 'STALE'

# Seed an OLD value, then wait so it falls outside the staleness window.
with database.batch() as batch:
    batch.insert_or_update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
        values=[[STALE_GROUP, STALE_ID, 100, spanner.COMMIT_TIMESTAMP]],
    )
print('Seeded OLD value = 100; waiting 16s so it ages past the 15s window...')
time.sleep(16)

# Now write a NEW value.
with database.batch() as batch:
    batch.insert_or_update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
        values=[[STALE_GROUP, STALE_ID, 999, spanner.COMMIT_TIMESTAMP]],
    )

q = f"""SELECT feature_int_1 FROM {SPANNER_TABLE}
        WHERE entity_group = '{STALE_GROUP}' AND entity_id = '{STALE_ID}'"""

# Strong read: linearizable, sees the latest committed write.
with database.snapshot(multi_use=True) as snapshot:
    strong_val = list(snapshot.execute_sql(q))[0][0]

# Exact-staleness read: served as of 15s ago -> before the new write.
with database.snapshot(exact_staleness=datetime.timedelta(seconds=15)) as snapshot:
    stale_val = list(snapshot.execute_sql(q))[0][0]

print(f'Wrote NEW value = 999')
print(f'  strong read           -> {strong_val}   (sees the new write)')
print(f'  exact_staleness=15s   -> {stale_val}   (still sees the old value)')
print()
print('Stale reads trade freshness for lower latency and replica-local serving.')
````

### Bounded snapshot: `read_timestamp` for reproducible reads

For reproducible reads — e.g. building a training set that must match a model's feature snapshot — pin the read to a specific commit timestamp. Every row is returned as of exactly that instant, regardless of later writes. This is Spanner's time-travel: any timestamp within the version-retention window (default 1 hour, configurable up to 7 days) is queryable.

````python
# Capture a timestamp 'now' (use the strong-read commit point as a reference).
snapshot_ts = datetime.datetime.now(datetime.timezone.utc)
print(f'Pinning reads to T = {snapshot_ts.isoformat()}')

# Mutate the scratch row AFTER our chosen timestamp.
with database.batch() as batch:
    batch.update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
        values=[[STALE_GROUP, STALE_ID, 12345, spanner.COMMIT_TIMESTAMP]],
    )

# Bounded snapshot at T does not see the post-T mutation.
with database.snapshot(read_timestamp=snapshot_ts) as snapshot:
    at_t = list(snapshot.execute_sql(q))[0][0]
with database.snapshot(multi_use=True) as snapshot:
    now_val = list(snapshot.execute_sql(q))[0][0]

print(f'  read_timestamp=T -> {at_t}   (reproducible: value as of T)')
print(f'  strong now       -> {now_val}   (reflects the later write)')

# Clean up both scratch rows used in this section.
with database.batch() as batch:
    batch.delete(table=SPANNER_TABLE,
                 keyset=spanner.KeySet(keys=[[STALE_GROUP, STALE_ID]]))
print('Cleaned up staleness scratch row.')
````

## 2b. Commit Timestamps Deep-Dive

Every write in this notebook sets `updated_at = spanner.COMMIT_TIMESTAMP` (or `PENDING_COMMIT_TIMESTAMP()` in DML). This is not a client clock value — it is a sentinel that tells Spanner to stamp the column with the **server-assigned commit timestamp** at the moment the transaction commits.

This works because the `updated_at` column was created with the option `allow_commit_timestamp=true`:

```sql
updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
```

Why it matters for a feature store:

- **Monotonicity / TrueTime.** Spanner derives commit timestamps from TrueTime and guarantees that if transaction T2 starts after T1 commits, then `commit_ts(T2) > commit_ts(T1)`. Timestamps are globally meaningful and never go backwards — even across regions.
- **Feature ordering.** `ORDER BY updated_at` gives the true write order of feature updates, with no dependence on client clocks (which can skew or jump). This is what makes "latest value wins" and freshness SLAs well-defined.
- **Time-travel alignment.** The commit timestamp is exactly the timestamp you can later pass to `read_timestamp=` (Section 2) for a reproducible snapshot. Stamp it, then replay the database as of that instant.

The constraint: a column with `allow_commit_timestamp=true` only accepts the commit-timestamp sentinel or values **at/behind** the current time — you cannot write a future timestamp into it.

````python
# Write three updates to scratch rows in sequence, each getting a fresh
# server-assigned commit timestamp, then read them back ordered by updated_at.
order_rows = [('Z', 'TS_A'), ('Z', 'TS_B'), ('Z', 'TS_C')]
for grp, eid in order_rows:
    with database.batch() as batch:
        batch.insert_or_update(
            table=SPANNER_TABLE,
            columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
            values=[[grp, eid, 1, spanner.COMMIT_TIMESTAMP]],
        )
    time.sleep(0.05)  # ensure distinct, increasing commit timestamps

with database.snapshot() as snapshot:
    ordered = list(snapshot.execute_sql(f"""
        SELECT entity_id, updated_at
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'Z' AND entity_id IN ('TS_A', 'TS_B', 'TS_C')
        ORDER BY updated_at
    """))

print('Rows ordered by server-assigned commit timestamp:')
prev = None
for eid, ts in ordered:
    monotonic = '' if prev is None else ('  (> previous)' if ts > prev else '  (!! not monotonic)')
    print(f'  {eid}: {ts.isoformat()}{monotonic}')
    prev = ts
print()
print('Order matches write order — commit timestamps are monotonic by TrueTime.')

# Clean up the timestamp-ordering scratch rows.
with database.batch() as batch:
    batch.delete(table=SPANNER_TABLE,
                 keyset=spanner.KeySet(keys=[list(r) for r in order_rows]))
print('Cleaned up commit-timestamp scratch rows.')
````

## 3. Batch DML — Multiple Updates in One Transaction

Execute multiple DML statements in a single transaction. They commit together atomically — all succeed or none do.

### The 80,000-mutation commit limit

A single Spanner read-write transaction can apply at most **80,000 mutations** per commit. A *mutation* is not a row — it is a single change to a single column value, and the count includes secondary-index maintenance. The counting rule:

- **Insert / Update**: counts **once per column affected**. The primary key columns are **always** counted (Spanner must locate/write the row), even if you only change one non-key column.
- **Delete**: counts as **1 per row deleted** (regardless of column count).
- **Secondary indexes**: **each index entry that is created, updated, or removed counts individually**. A write touching an indexed column therefore costs extra mutations — one per affected index per row.

> Worked example: updating 3 non-key columns on 1,000 rows in a table with one secondary index on one of those columns =
> `1000 rows × (1 PK + 3 columns)` for the base table `+ 1000` index-entry updates `= 5,000 mutations`.

`batch_update` groups multiple DML statements into one round trip *and* one atomic commit, but the combined work still counts against the same 80,000-mutation budget. For larger writes, use **Partitioned DML** (Section 4) or chunk the work across multiple commits.

````python
def batch_dml_update(transaction):
    """Update multiple entities with batch DML.

    transaction.batch_update accepts plain SQL strings, or
    (sql, params, param_types) tuples for parameterized statements.
    It issues all statements in one round trip and returns
    (status, [row_count_per_statement]).
    """
    status, row_counts = transaction.batch_update([
        f"UPDATE {SPANNER_TABLE} SET feature_float_1 = feature_float_1 + 0.1, "
        f"updated_at = PENDING_COMMIT_TIMESTAMP() "
        f"WHERE entity_group = 'A' AND entity_id = '000010'",

        f"UPDATE {SPANNER_TABLE} SET feature_float_1 = feature_float_1 + 0.2, "
        f"updated_at = PENDING_COMMIT_TIMESTAMP() "
        f"WHERE entity_group = 'A' AND entity_id = '000020'",

        f"UPDATE {SPANNER_TABLE} SET feature_float_1 = feature_float_1 + 0.3, "
        f"updated_at = PENDING_COMMIT_TIMESTAMP() "
        f"WHERE entity_group = 'A' AND entity_id = '000030'",
    ])
    return status, row_counts

start = time.perf_counter()
status, row_counts = database.run_in_transaction(batch_dml_update)
batch_dml_ms = (time.perf_counter() - start) * 1000

print(f'Batch DML (3 statements): {batch_dml_ms:.1f} ms')
print(f'  Rows updated per statement: {row_counts}')
print(f'\nAll 3 updates committed atomically — or none would have.')

# A parameterized variant uses (sql, params, param_types) tuples:
def batch_dml_parameterized(transaction):
    status, row_counts = transaction.batch_update([
        (
            f"UPDATE {SPANNER_TABLE} SET feature_float_1 = feature_float_1 + @delta, "
            f"updated_at = PENDING_COMMIT_TIMESTAMP() "
            f"WHERE entity_group = @grp AND entity_id = @eid",
            {'delta': 0.5, 'grp': 'A', 'eid': '000010'},
            {'delta': spanner.param_types.FLOAT64,
             'grp': spanner.param_types.STRING,
             'eid': spanner.param_types.STRING},
        ),
    ])
    return row_counts

pcounts = database.run_in_transaction(batch_dml_parameterized)
print(f'Parameterized batch DML rows updated: {pcounts}')
````

## 3b. Partitioned DML — Large-Scale Backfills

The 80,000-mutation commit limit makes a single transaction unsuitable for backfilling a feature across millions of rows. **Partitioned DML** solves this: Spanner splits the table into partitions and runs the statement independently against each one.

```python
database.execute_partitioned_dml(
    "UPDATE dense_features SET feature_float_3 = 0.0 WHERE feature_float_3 IS NULL"
)
```

Key properties:

- **Bypasses the 80k-mutation limit** — there is no single giant commit; each partition commits on its own.
- **Weaker atomicity.** Atomicity is **per-partition, not global**. If the operation fails midway, some partitions may be committed and others not. Partitioned DML is therefore only safe for statements that are **idempotent** (re-running produces the same result) — exactly the shape of feature backfills like "set default where null" or "rescale a column".
- **Single statement, no reads of buffered writes**, and the statement must be **partitionable** (the WHERE must let Spanner route rows to partitions independently).

Below we run a deliberately **no-op-safe** statement: set `feature_float_3` to itself where it is null. It exercises the partitioned path without changing any data.

````python
# No-op-safe partitioned DML: assign the column to itself.
# (Idempotent and data-preserving — safe to run on live data.)
start = time.perf_counter()
affected = database.execute_partitioned_dml(
    f"UPDATE {SPANNER_TABLE} SET feature_float_3 = feature_float_3 "
    f"WHERE feature_float_3 IS NULL"
)
pdml_ms = (time.perf_counter() - start) * 1000

print(f'Partitioned DML completed in {pdml_ms:.0f} ms')
print(f'  Rows affected (lower-bound count): {affected}')
print()
print('No single commit — Spanner partitioned the work, so the 80k-mutation')
print('commit limit never applied. Atomicity was per-partition, not global.')
````

### Regular DML vs Batch DML vs Partitioned DML

| | Regular DML / `transaction.update` | Batch DML (`batch_update`) | Partitioned DML |
|---|---|---|---|
| **Statements per call** | 1 | Many (one round trip) | 1 |
| **Atomicity scope** | Global (whole transaction) | Global (all statements commit together) | **Per-partition** (not global) |
| **Mutation limit** | 80,000 per commit | 80,000 per commit (shared budget) | **None** (partitioned across many commits) |
| **Reads in same txn** | Yes | Yes | No |
| **Idempotency required** | No | No | **Yes** (may partially apply on failure) |
| **Feature-store use case** | Real-time read-modify-write of a few rows | Several related updates committed together | **Backfills / bulk rescale / default-fill across millions of rows** |

## 3c. The Mutation Limit, Made Concrete

Recall the counting rule from Section 3: a mutation is one column-value change, the primary key is always counted, deletes count 1/row, and **each secondary-index entry change counts on its own**. Let's compute how big a single-commit batch can get for this table, then show what happens at the boundary and how to work around it.

````python
# Count columns on dense_features and any secondary indexes on it.
with database.snapshot(multi_use=True) as snapshot:
    n_columns = list(snapshot.execute_sql(f"""
        SELECT COUNT(*) FROM INFORMATION_SCHEMA.COLUMNS
        WHERE TABLE_NAME = '{SPANNER_TABLE}' AND TABLE_SCHEMA = ''
    """))[0][0]
    n_indexes = list(snapshot.execute_sql(f"""
        SELECT COUNT(*) FROM INFORMATION_SCHEMA.INDEXES
        WHERE TABLE_NAME = '{SPANNER_TABLE}' AND TABLE_SCHEMA = ''
          AND INDEX_TYPE = 'INDEX'
    """))[0][0]

MUTATION_LIMIT = 80_000

# Worst case for a full-row insert: every column + one entry per secondary index.
mutations_per_row = n_columns + n_indexes
max_rows_per_commit = MUTATION_LIMIT // mutations_per_row

print(f'{SPANNER_TABLE}: {n_columns} columns, {n_indexes} secondary index(es)')
print(f'Full-row insert cost: {n_columns} (columns, PK included) '
      f'+ {n_indexes} (index entries) = {mutations_per_row} mutations/row')
print(f'Max rows in ONE commit: {MUTATION_LIMIT:,} // {mutations_per_row} '
      f'= {max_rows_per_commit:,} rows')
print()
print('Worked example (independent of this table):')
print('  Update 3 non-key columns on 1,000 rows, 1 secondary index on a touched column')
print('  = 1000 x (1 PK + 3 cols) + 1000 index entries = 5,000 mutations.')
````

### Crossing the boundary

Exceeding 80,000 mutations in one commit fails with `FailedPrecondition: The transaction contains too many mutations`. Rather than insert tens of thousands of throwaway rows on a live database, we demonstrate the limit *defensively*: we compute a row count that would exceed it and show the two safe workarounds — **chunk** the work across multiple commits, or use **Partitioned DML**.

````python
# How many rows would blow the budget?
rows_that_exceed = max_rows_per_commit + 1
print(f'Inserting {rows_that_exceed:,} full rows in one commit would require '
      f'{rows_that_exceed * mutations_per_row:,} mutations (> {MUTATION_LIMIT:,}).')
print('That single commit would raise:')
print('  google.api_core.exceptions.FailedPrecondition:')
print('  "The transaction contains too many mutations."')
print()

# Workaround 1: chunk into multiple commits, each under the limit.
CHUNK = max_rows_per_commit  # safely <= limit
total = rows_that_exceed
n_chunks = (total + CHUNK - 1) // CHUNK
print(f'Workaround A - chunked batches: {total:,} rows / {CHUNK:,} per commit '
      f'= {n_chunks} commits, each within the 80k budget.')
print('  for start in range(0, total, CHUNK):')
print('      with database.batch() as batch:')
print('          batch.insert_or_update(... rows[start:start+CHUNK] ...)')
print()

# Workaround 2: Partitioned DML for set-based backfills (Section 3b).
print('Workaround B - partitioned DML: for a set-based change ')
print('  (e.g. default-fill or rescale), execute_partitioned_dml has no')
print('  mutation limit because it never builds one giant commit.')
````

## 4. Cross-Table Transactions

Update rows across different tables in the same transaction. This is critical for multi-entity feature stores where an action affects multiple entity types (e.g., a user clicks an item → update both user and item features).

````python
# Check if interleaved tables exist (from NB4)
with database.snapshot() as snapshot:
    tables = [row[0] for row in snapshot.execute_sql(
        "SELECT TABLE_NAME FROM INFORMATION_SCHEMA.TABLES WHERE TABLE_SCHEMA = ''"
    )]

has_interleaved = 'entity_features' in tables and 'entity_groups' in tables
print(f'Interleaved tables available: {has_interleaved}')
print(f'Tables: {tables}')
````

````python
if has_interleaved:
    def cross_table_update(transaction):
        """Update entity features and group aggregates atomically."""
        # Read current entity value
        entity = list(transaction.execute_sql("""
            SELECT feature_float_1 FROM entity_features
            WHERE entity_group = 'A' AND entity_id = '000001'
        """))[0]

        new_float = entity[0] + 1.0

        # Update entity features
        transaction.update(
            table='entity_features',
            columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
            values=[['A', '000001', new_float, spanner.COMMIT_TIMESTAMP]],
        )

        # Recalculate group aggregate
        avg = list(transaction.execute_sql("""
            SELECT AVG(feature_float_1) FROM entity_features
            WHERE entity_group = 'A'
        """))[0][0]

        # Update group-level feature
        transaction.update(
            table='entity_groups',
            columns=['entity_group', 'group_avg_float_1', 'updated_at'],
            values=[['A', avg, spanner.COMMIT_TIMESTAMP]],
        )

        return entity[0], new_float, avg

    start = time.perf_counter()
    old, new, avg = database.run_in_transaction(cross_table_update)
    cross_ms = (time.perf_counter() - start) * 1000

    print(f'Cross-table transaction: {cross_ms:.1f} ms')
    print(f'  entity_features A/000001: float_1 = {old:.2f} → {new:.2f}')
    print(f'  entity_groups A: avg_float_1 = {avg:.2f}')
    print(f'\nBoth tables updated atomically — impossible in Bigtable.')
else:
    print('Skipping cross-table demo (run NB4: Interleaved Tables first).')
````

## 5. Transaction Latency Benchmark

Compare latency across transaction types.

````python
# Single-row batch write (no transaction overhead)
batch_write_lats = []
for i in range(20):
    eid = f'{(i % 100) + 1:06d}'
    start = time.perf_counter()
    with database.batch() as batch:
        batch.update(
            table=SPANNER_TABLE,
            columns=['entity_group', 'entity_id', 'feature_float_2', 'updated_at'],
            values=[['B', eid, 50.0 + i, spanner.COMMIT_TIMESTAMP]],
        )
    batch_write_lats.append((time.perf_counter() - start) * 1000)

# Single-row read-write transaction
rw_single_lats = []
for i in range(20):
    eid = f'{(i % 100) + 1:06d}'
    def rw_single(transaction, eid=eid):
        result = list(transaction.execute_sql(f"""
            SELECT feature_int_1 FROM {SPANNER_TABLE}
            WHERE entity_group = 'B' AND entity_id = '{eid}'
        """))
        transaction.update(
            table=SPANNER_TABLE,
            columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
            values=[['B', eid, result[0][0] + 1, spanner.COMMIT_TIMESTAMP]],
        )
    start = time.perf_counter()
    database.run_in_transaction(rw_single)
    rw_single_lats.append((time.perf_counter() - start) * 1000)

# Multi-row read-write transaction
rw_multi_lats = []
for i in range(20):
    def rw_multi(transaction, offset=i):
        eids = [f'{(offset * 3 + j) % 100 + 1:06d}' for j in range(3)]
        placeholders = ', '.join(f"'{e}'" for e in eids)
        results = list(transaction.execute_sql(f"""
            SELECT entity_id, feature_int_1 FROM {SPANNER_TABLE}
            WHERE entity_group = 'B' AND entity_id IN ({placeholders})
        """))
        values = [['B', r[0], r[1] + 1, spanner.COMMIT_TIMESTAMP] for r in results]
        transaction.update(
            table=SPANNER_TABLE,
            columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
            values=values,
        )
    start = time.perf_counter()
    database.run_in_transaction(rw_multi)
    rw_multi_lats.append((time.perf_counter() - start) * 1000)

print(f'Transaction latency comparison (20 iterations each):\n')
print(f'{"Type":<35s} {"Median":>10s} {"p95":>10s}')
for name, lats in [('Batch write (no txn)', batch_write_lats),
                    ('R/W txn (1 row)', rw_single_lats),
                    ('R/W txn (3 rows)', rw_multi_lats)]:
    s = sorted(lats)
    print(f'{name:<35s} {s[10]:>8.1f}ms {s[19]:>8.1f}ms')
````

````python
fig, ax = plt.subplots(figsize=(10, 5))

data = [batch_write_lats, rw_single_lats, rw_multi_lats]
labels = ['Batch write\n(no txn)', 'R/W txn\n(1 row)', 'R/W txn\n(3 rows)']
colors = ['#34A853', '#FBBC04', '#EA4335']

bp = ax.boxplot(data, tick_labels=labels, patch_artist=True)
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('Latency (ms)', fontweight='bold')
ax.set_title('Write Latency by Transaction Type', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
````

## 5b. Lock Contention and Tail Latency

The benchmark above measured uncontended transactions. In production, the cost that bites is **lock contention**: when many writers target the same hot row, Spanner serializes them, and the waiters' latency inflates the **tail** (p95/p99) even though the median stays low.

Here we measure the same single-row read-write transaction two ways:

1. **Uncontended** — each iteration writes a *different* row, so writers never wait on each other.
2. **Hot row** — background threads continuously hammer one row while we time transactions against that *same* row. The measured transactions must wait for locks, so p95/p99 climb.

We use the reserved scratch row for the hot path so no real feature data is contended.

````python
import statistics

def pctl(values, p):
    s = sorted(values)
    k = min(len(s) - 1, int(round((p / 100) * (len(s) - 1))))
    return s[k]

def time_rw_on(grp, eid):
    """One read-modify-write txn against (grp, eid); returns latency ms."""
    def fn(transaction):
        row = list(transaction.execute_sql(f"""
            SELECT feature_int_1 FROM {SPANNER_TABLE}
            WHERE entity_group = '{grp}' AND entity_id = '{eid}'
        """))
        cur = row[0][0] if row else 0
        transaction.update(
            table=SPANNER_TABLE,
            columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
            values=[[grp, eid, cur + 1, spanner.COMMIT_TIMESTAMP]],
        )
    t = time.perf_counter()
    database.run_in_transaction(fn)
    return (time.perf_counter() - t) * 1000

N_MEASURE = 40

# --- 1. Uncontended: each iteration hits a different existing row ---
uncontended = [time_rw_on('B', f'{(i % 100) + 1:06d}') for i in range(N_MEASURE)]

# --- 2. Hot row: background writers contend on the scratch row ---
reset_scratch(0)
stop_flag = {'stop': False}
def hammer():
    while not stop_flag['stop']:
        try:
            time_rw_on(SCRATCH_GROUP, SCRATCH_ID)
        except Exception:
            pass

contended = []
with ThreadPoolExecutor(max_workers=6) as pool:
    bg = [pool.submit(hammer) for _ in range(6)]
    for _ in range(N_MEASURE):
        contended.append(time_rw_on(SCRATCH_GROUP, SCRATCH_ID))
    stop_flag['stop'] = True

print(f'RW-txn latency: uncontended row vs hot row ({N_MEASURE} samples each)\n')
print(f'{"Scenario":<22s} {"p50":>9s} {"p95":>9s} {"p99":>9s}')
for name, lats in [('Uncontended row', uncontended), ('Hot row (6 writers)', contended)]:
    print(f'{name:<22s} {pctl(lats,50):>7.1f}ms {pctl(lats,95):>7.1f}ms {pctl(lats,99):>7.1f}ms')
print()
print('Lock waits on the hot row inflate p95/p99 while p50 stays modest —')
print('classic tail-latency signature of write contention on a single key.')
````

````python
fig, ax = plt.subplots(figsize=(10, 5))
data = [uncontended, contended]
labels = ['Uncontended\nrow', 'Hot row\n(6 writers)']
colors = ['#34A853', '#EA4335']
bp = ax.boxplot(data, tick_labels=labels, patch_artist=True, showfliers=True)
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color); patch.set_alpha(0.7)
ax.set_ylabel('RW-txn latency (ms)', fontweight='bold')
ax.set_title('Lock Contention Inflates Tail Latency', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
````

## 6. Comparison to Bigtable Atomic Operations

| Capability | Spanner | Bigtable |
|---|---|---|
| **Scope** | Multi-row, multi-table | Single-row only |
| **Read-modify-write** | Read → compute → write in transaction | `ReadModifyWriteRow` (append/increment only) |
| **Conditional update** | SQL WHERE in DML | `CheckAndMutateRow` (row filter → true/false mutation) |
| **Batch update** | Up to 80,000 mutations / commit | `MutateRows` (no read, no atomicity across rows) |
| **Staleness** | Configurable (strong, exact, max) | Always strong (single cluster) |
| **Latency** | ~5–15ms (transaction overhead) | ~3–5ms (single-row, no transaction) |
| **Conflicts** | Automatic retry with backoff | N/A (single-row = no conflicts) |

**When Spanner transactions matter for feature stores:**
- Update user features + item features atomically on a click event
- Maintain consistent aggregates (group-level stats must reflect entity-level changes)
- Conditional updates: only update if the feature version matches (optimistic concurrency)
- Transfer-style operations: debit one entity, credit another

---
## Summary

| Transaction Type | Latency | Scope |
|---|---|---|
| Batch write (no transaction) | ~3–8ms | Single-row or multi-row (no read) |
| Read-write (1 row) | ~5–12ms | Read + write atomically |
| Read-write (3 rows) | ~8–18ms | Multi-row atomic |
| Cross-table | ~10–20ms | Parent + child tables |
| Read-only (strong) | ~5–10ms | Linearizable consistency |
| Read-only (stale 10s) | ~3–7ms | Lower latency, bounded staleness |

### Key concepts

- **Serializability with auto-retry.** `run_in_transaction` re-runs your function on abort. Correct read-modify-write loops never lose updates under concurrency — they pay in retries.
- **Staleness is a freshness trade-off**, not just a latency knob: strong reads see the latest write; `exact_staleness` deliberately sees older data; `read_timestamp` pins a reproducible snapshot.
- **Commit timestamps** are server-assigned and monotonic (TrueTime), giving well-defined feature ordering and time-travel alignment.
- **80,000 mutations per commit**, counting per-column-per-row plus every secondary-index entry. Use **chunked batches** or **Partitioned DML** to go bigger.
- **Partitioned DML** bypasses the limit for idempotent backfills, with per-partition (not global) atomicity.
- **Lock contention** inflates p95/p99 on hot rows even when p50 is fine.
