# Spanner Feature Store — Schema Evolution

Add, drop, and modify feature columns without downtime using [online schema changes](https://cloud.google.com/spanner/docs/schema-updates). Spanner DDL runs in the background while reads and writes continue.

**What you'll learn:**
- Online schema change: `ALTER TABLE ADD COLUMN` with zero downtime
- Non-blocking DDL: *prove* reads and writes continue *during* a schema change
- Backfill new columns with partitioned DML (and why commit-timestamp columns need a separate transaction)
- Drop columns: `ALTER TABLE DROP COLUMN`
- `DEFAULT`-value columns: existing rows read the default immediately
- Generated (`STORED`) columns — a Spanner-unique feature with no Bigtable/Valkey analog
- The canonical type-change recipe: add new column → backfill with `CAST` → drop old column
- What schema changes are allowed vs. forbidden (compatibility matrix)
- DDL and backfill timing (median / p95)
- Schema version tracking via a metadata table + native operation history
- `INFORMATION_SCHEMA` for runtime schema discovery and forward-compatible readers

**Key difference from Bigtable:** Bigtable is schema-free — you can write any column qualifier at any time without DDL. This is flexible but provides no type enforcement. Spanner is strongly typed — DDL is required but executes online, and type errors are caught at write time.

**Prerequisites:** Run [Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) and [Fundamentals](./Spanner%20Feature%20Store%20-%20Fundamentals.ipynb) first.

```python
import os  # noqa
# Silence the Spanner client's built-in Cloud Monitoring metrics exporter
# (noisy on incomplete resource labels; unrelated to the notebook).
os.environ.setdefault("SPANNER_DISABLE_BUILTIN_METRICS", "true")
from google.cloud import spanner
import time

spanner_client = spanner.Client(project=PROJECT_ID)
instance = spanner_client.instance(SPANNER_INSTANCE)
database = instance.database(SPANNER_DATABASE)

print(f'Spanner: {SPANNER_INSTANCE}/{SPANNER_DATABASE}')

def column_exists(col, table=None):
    """True if `col` exists on the table (Spanner has no DROP COLUMN IF EXISTS)."""
    table = table or SPANNER_TABLE
    with database.snapshot() as snap:
        rows = list(snap.execute_sql(
            'SELECT 1 FROM INFORMATION_SCHEMA.COLUMNS '
            'WHERE TABLE_NAME = @t AND COLUMN_NAME = @c',
            params={'t': table, 'c': col},
            param_types={'t': spanner.param_types.STRING, 'c': spanner.param_types.STRING}))
    return len(rows) > 0

def drop_column(col, table=None):
    """Idempotent column drop: Spanner DDL lacks DROP COLUMN IF EXISTS, so we guard."""
    table = table or SPANNER_TABLE
    if column_exists(col, table):
        database.update_ddl([f'ALTER TABLE {table} DROP COLUMN {col}']).result()
        return True
    return False
```

## 1. Current Schema — Baseline

Before making changes, capture the current schema from `INFORMATION_SCHEMA`.

```python
def get_schema():
    with database.snapshot() as snapshot:
        columns = list(snapshot.execute_sql("""
            SELECT COLUMN_NAME, SPANNER_TYPE, IS_NULLABLE
            FROM INFORMATION_SCHEMA.COLUMNS
            WHERE TABLE_NAME = @table
            ORDER BY ORDINAL_POSITION
        """, params={'table': SPANNER_TABLE}, param_types={'table': spanner.param_types.STRING}))
    return columns

schema_before = get_schema()
print(f'Current schema: {len(schema_before)} columns')
print(f'\nFirst 10:')
for name, stype, nullable in schema_before[:10]:
    print(f'  {name:<35s} {stype:<20s} nullable={nullable}')
print(f'  ...')
print(f'\nLast 5:')
for name, stype, nullable in schema_before[-5:]:
    print(f'  {name:<35s} {stype:<20s} nullable={nullable}')
```

## 2. Add New Feature Columns — Online DDL

`ALTER TABLE ADD COLUMN` executes in the background. Existing reads and writes continue uninterrupted. New columns are initially NULL for all existing rows.

```python
# Add new feature columns
add_columns_ddl = [
    f'ALTER TABLE {SPANNER_TABLE} ADD COLUMN IF NOT EXISTS feature_v2_score FLOAT64',
    f'ALTER TABLE {SPANNER_TABLE} ADD COLUMN IF NOT EXISTS feature_v2_category STRING(50)',
    f'ALTER TABLE {SPANNER_TABLE} ADD COLUMN IF NOT EXISTS feature_v2_embedding ARRAY<FLOAT64>',
]

start = time.perf_counter()
try:
    operation = database.update_ddl(add_columns_ddl)
    operation.result()
    ddl_ms = (time.perf_counter() - start) * 1000
    print(f'Added 3 columns in {ddl_ms:.0f} ms')
except Exception as e:
    ddl_ms = (time.perf_counter() - start) * 1000
    print(f'DDL completed in {ddl_ms:.0f} ms: {e}')

# Verify new columns exist
schema_after = get_schema()
new_cols = [c for c in schema_after if c[0].startswith('feature_v2_')]
print(f'\nNew columns:')
for name, stype, nullable in new_cols:
    print(f'  {name:<35s} {stype:<20s} nullable={nullable}')
```

```python
# Verify reads still work during/after schema change
with database.snapshot() as snapshot:
    result = list(snapshot.execute_sql(f"""
        SELECT entity_group, entity_id, feature_float_1, feature_v2_score, feature_v2_category
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id = '000001'
    """))

row = result[0]
print(f'Read after schema change:')
print(f'  entity: {row[0]}/{row[1]}')
print(f'  feature_float_1:    {row[2]}')
print(f'  feature_v2_score:   {row[3]} (NULL — new column, not yet populated)')
print(f'  feature_v2_category: {row[4]} (NULL)')
```

### Proof: reads and writes succeed *during* the schema change

The headline claim of online DDL is that traffic keeps flowing while the schema changes. Let's *prove* it instead of asserting it: kick off an `ADD COLUMN`/`DROP COLUMN` operation and, while `op.done()` is still `False`, hammer the table with concurrent reads and writes from a thread pool. Every one of them should succeed.

```python
import concurrent.futures

# A column we will add and immediately drop, purely to have a DDL operation to
# observe. Using a fresh name so it never collides with the v2 demo columns.
PROBE_COL = 'feature_ddl_probe'

stop_flag = {'stop': False}
results = {'reads': 0, 'writes': 0, 'errors': 0}

def hammer_reads():
    n = 0
    while not stop_flag['stop']:
        try:
            with database.snapshot() as snap:
                list(snap.execute_sql(
                    f"SELECT feature_float_1 FROM {SPANNER_TABLE} "
                    f"WHERE entity_group = 'A' AND entity_id = '000002'"))
            n += 1
        except Exception:
            results['errors'] += 1
    results['reads'] = n

def hammer_writes():
    n = 0
    while not stop_flag['stop']:
        try:
            def _w(txn, val=n):
                txn.execute_update(
                    f"UPDATE {SPANNER_TABLE} SET feature_int_2 = @v "
                    f"WHERE entity_group = 'A' AND entity_id = '000002'",
                    params={'v': 1000 + (val % 100)},
                    param_types={'v': spanner.param_types.INT64})
            database.run_in_transaction(_w)
            n += 1
        except Exception:
            results['errors'] += 1
    results['writes'] = n

with concurrent.futures.ThreadPoolExecutor(max_workers=4) as pool:
    readers = [pool.submit(hammer_reads) for _ in range(2)]
    writers = [pool.submit(hammer_writes) for _ in range(2)]

    # Kick off the DDL and poll op.done() while traffic flows.
    op = database.update_ddl([f'ALTER TABLE {SPANNER_TABLE} ADD COLUMN IF NOT EXISTS {PROBE_COL} FLOAT64'])
    polls = 0
    t0 = time.perf_counter()
    while not op.done():
        polls += 1
        time.sleep(0.05)
    op.result()  # surface any DDL error
    ddl_observed_ms = (time.perf_counter() - t0) * 1000

    stop_flag['stop'] = True
    concurrent.futures.wait(readers + writers)

print(f'ADD COLUMN observed running for ~{ddl_observed_ms:.0f} ms across {polls} op.done() poll(s).')
print(f'During the DDL: {results["reads"]:,} reads and {results["writes"]:,} writes succeeded, '
      f'{results["errors"]} errors.')
if polls <= 1:
    print('(DDL completed almost immediately at this small scale — the pattern is what '
          'matters: traffic never blocks on the schema operation.)')

# Clean up the probe column so the table schema stays sane.
drop_column(PROBE_COL)
print(f'Dropped {PROBE_COL}. Reads/writes saw zero downtime throughout.')
```

## 3. Backfill New Columns

After adding columns, populate them with computed values. Use partitioned DML for large-scale backfills — it breaks the update into smaller transactions automatically.

```python
# Backfill feature_v2_score / feature_v2_category as a function of existing features.
#
# IMPORTANT: partitioned DML CANNOT set a commit-timestamp column
# (PENDING_COMMIT_TIMESTAMP() is rejected — partitioned DML runs as many
# independent transactions, so there is no single commit timestamp to pin).
# So we backfill the VALUES with partitioned DML (scales to millions of rows),
# then bump updated_at separately in a regular read-write transaction.
start = time.perf_counter()
row_count = database.execute_partitioned_dml(f"""
    UPDATE {SPANNER_TABLE}
    SET feature_v2_score = (feature_float_1 * 0.4 + feature_float_2 * 0.3 + CAST(feature_int_1 AS FLOAT64) * 0.003),
        feature_v2_category = CASE
            WHEN feature_float_1 > 75 THEN 'high'
            WHEN feature_float_1 > 25 THEN 'medium'
            ELSE 'low'
        END
    WHERE feature_v2_score IS NULL
""")
backfill_ms = (time.perf_counter() - start) * 1000

print(f'Backfill complete: {row_count:,} rows in {backfill_ms:.0f} ms')
print(f'  Used partitioned DML — automatically splits into smaller transactions.')
print(f'  (No commit-timestamp column in this UPDATE — partitioned DML forbids it.)')

# Touch updated_at for one representative entity in a REGULAR transaction, where
# PENDING_COMMIT_TIMESTAMP() IS allowed. For a full-table audit-timestamp bump
# you would loop batched key ranges through run_in_transaction the same way.
def _stamp_updated_at(transaction):
    return transaction.execute_update(
        f"""UPDATE {SPANNER_TABLE}
            SET updated_at = PENDING_COMMIT_TIMESTAMP()
            WHERE entity_group = @g AND entity_id = @e""",
        params={'g': 'A', 'e': '000001'},
        param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING},
    )

stamped = database.run_in_transaction(_stamp_updated_at)
print(f'Stamped updated_at on {stamped} row via a regular DML transaction (commit timestamp allowed there).')
```

```python
# Verify backfill
with database.snapshot(multi_use=True) as snapshot:
    sample = list(snapshot.execute_sql(f"""
        SELECT entity_group, entity_id, feature_float_1, feature_v2_score, feature_v2_category
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A'
        ORDER BY entity_id
        LIMIT 5
    """))

    null_count = list(snapshot.execute_sql(f"""
        SELECT COUNTIF(feature_v2_score IS NULL) AS null_count, COUNT(*) AS total
        FROM {SPANNER_TABLE}
    """))[0]

print(f'Sample backfilled rows:')
print(f'{"Entity":>12s} {"float_1":>10s} {"v2_score":>10s} {"v2_category":>12s}')
for row in sample:
    print(f'{row[0]}/{row[1]:>8s} {row[2]:>10.2f} {row[3]:>10.2f} {row[4]:>12s}')

print(f'\nRemaining NULLs: {null_count[0]:,} / {null_count[1]:,}')
```

## 3a. `DEFAULT`-Value Columns

The intro promised default values; here is the demo. When you `ADD COLUMN ... DEFAULT (expr)`, **existing rows immediately read back the default** instead of `NULL`. The default is stored as column metadata, so there is no backfill to wait for — contrast this with section 2, where the new `feature_v2_*` columns read `NULL` until we backfilled them.

New rows that don't specify the column also get the default; rows that *do* specify it use the supplied value.

```python
# Add a column WITH a default. Existing rows read the default immediately — no backfill.
database.update_ddl([
    f'ALTER TABLE {SPANNER_TABLE} ADD COLUMN IF NOT EXISTS feature_v2_default FLOAT64 DEFAULT (0.0)'
]).result()
print('Added feature_v2_default FLOAT64 DEFAULT (0.0).')

with database.snapshot(multi_use=True) as snapshot:
    existing = list(snapshot.execute_sql(f"""
        SELECT entity_group, entity_id, feature_v2_default
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id = '000001'
    """))[0]
    null_check = list(snapshot.execute_sql(f"""
        SELECT COUNTIF(feature_v2_default IS NULL) AS null_count, COUNT(*) AS total
        FROM {SPANNER_TABLE}
    """))[0]

print(f'Existing row reads back: feature_v2_default = {existing[2]}  '
      f'(the DEFAULT, not NULL — no backfill required)')
print(f'NULLs across the table: {null_check[0]:,} / {null_check[1]:,}  '
      f'(contrast with feature_v2_* which were NULL until backfilled)')

# A new row that omits the column also gets the default; one that sets it overrides.
# (updated_at is NOT NULL with allow_commit_timestamp, so we must set it on insert.)
def _insert_default_demo(txn):
    txn.execute_update(
        f"INSERT INTO {SPANNER_TABLE} (entity_group, entity_id, updated_at) "
        f"VALUES ('Z', 'dflt01', PENDING_COMMIT_TIMESTAMP())")
    txn.execute_update(
        f"INSERT INTO {SPANNER_TABLE} (entity_group, entity_id, feature_v2_default, updated_at) "
        f"VALUES ('Z', 'dflt02', 9.9, PENDING_COMMIT_TIMESTAMP())")
database.run_in_transaction(_insert_default_demo)

with database.snapshot(multi_use=True) as snapshot:
    rows = list(snapshot.execute_sql(f"""
        SELECT entity_id, feature_v2_default FROM {SPANNER_TABLE}
        WHERE entity_group = 'Z' AND entity_id IN ('dflt01', 'dflt02') ORDER BY entity_id
    """))
for eid, val in rows:
    print(f'  new row {eid}: feature_v2_default = {val}'
          f'{"  (omitted → default)" if eid == "dflt01" else "  (explicit value)"}')
```

## 3b. Generated (`STORED`) Columns — Spanner-Unique

A **generated column** is computed from other columns in the same row by an expression you define in DDL. With `STORED`, the value is materialized on write and kept consistent automatically — you never set it, and you can't write to it directly. Bigtable and Valkey have **no analog**: there, derived features are the application's job.

This makes generated columns ideal for derived features (ratios, bucketizations, normalizations) that must stay in lockstep with their inputs.

```python
# feature_ratio = feature_float_1 / feature_float_2, guarded against divide-by-zero.
# STORED → materialized on write and recomputed automatically on every update.
database.update_ddl([
    f'ALTER TABLE {SPANNER_TABLE} ADD COLUMN IF NOT EXISTS '
    f'feature_ratio FLOAT64 AS (feature_float_1 / NULLIF(feature_float_2, 0)) STORED'
]).result()
print('Added generated column feature_ratio (STORED).')

# Pick an entity with non-null, non-zero inputs so the ratio is well-defined.
with database.snapshot(multi_use=True) as snapshot:
    pick = list(snapshot.execute_sql(f"""
        SELECT entity_group, entity_id, feature_float_1, feature_float_2, feature_ratio
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND feature_float_1 IS NOT NULL
          AND feature_float_2 IS NOT NULL AND feature_float_2 != 0
        ORDER BY entity_id LIMIT 1
    """))[0]
g, e, f1, f2, ratio = pick
print(f'Auto-populated on {g}/{e}: {f1:.3f} / {f2:.3f} = feature_ratio {ratio:.4f}')

# Now CHANGE an input and confirm the generated column tracks it — we update only
# feature_float_1; feature_ratio recomputes by itself (we cannot set it directly).
def _bump_input(txn):
    txn.execute_update(
        f"UPDATE {SPANNER_TABLE} SET feature_float_1 = feature_float_1 + 10 "
        f"WHERE entity_group = @g AND entity_id = @e",
        params={'g': g, 'e': e},
        param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING})
database.run_in_transaction(_bump_input)

with database.snapshot(multi_use=True) as snapshot:
    after = list(snapshot.execute_sql(f"""
        SELECT feature_float_1, feature_float_2, feature_ratio
        FROM {SPANNER_TABLE} WHERE entity_group = @g AND entity_id = @e
    """, params={'g': g, 'e': e},
         param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING}))[0]
expected = after[0] / after[1]
print(f'After bumping feature_float_1: {after[0]:.3f} / {after[1]:.3f} = '
      f'feature_ratio {after[2]:.4f} (expected {expected:.4f}) — stayed consistent.')

# Confirm you cannot write to a generated column directly.
try:
    def _illegal(txn):
        txn.execute_update(
            f"UPDATE {SPANNER_TABLE} SET feature_ratio = 1.0 "
            f"WHERE entity_group = @g AND entity_id = @e",
            params={'g': g, 'e': e},
            param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING})
    database.run_in_transaction(_illegal)
    print('Unexpected: direct write to generated column succeeded.')
except Exception as ex:
    print(f'Direct write to feature_ratio rejected (expected): {type(ex).__name__}')

# restore the input we bumped, so later notebooks see the original value
database.run_in_transaction(lambda txn: txn.execute_update(
    f"UPDATE {SPANNER_TABLE} SET feature_float_1 = feature_float_1 - 10 "
    f"WHERE entity_group = @g AND entity_id = @e",
    params={'g': g, 'e': e},
    param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING}))
print('Restored feature_float_1.')
```

## 3c. Changing a Column's Type — the Add/Backfill/Drop Recipe

Spanner **cannot** `ALTER` a column across incompatible types (e.g. `STRING` → `INT64`). The canonical, zero-downtime pattern is:

1. **Add** a new column of the target type (online, instant).
2. **Backfill** it from the old column with `CAST` / `CASE` via partitioned DML.
3. **Cut over** readers/writers to the new column.
4. **Drop** the old column once nothing references it.

Below we demonstrate it safely by deriving an `INT64` code from the `feature_v2_category` string we backfilled earlier (`low`/`medium`/`high` → `0`/`1`/`2`), then cleaning up the throwaway column.

```python
# Step 0: prove the in-place type change is forbidden (so the recipe is justified).
try:
    database.update_ddl([
        f'ALTER TABLE {SPANNER_TABLE} ALTER COLUMN feature_v2_category INT64'
    ]).result()
    print('Unexpected: in-place STRING→INT64 succeeded.')
except Exception as e:
    print(f'In-place STRING→INT64 rejected (expected): {str(e).splitlines()[0][:120]}')

# Step 1: add the new-typed column (online).
database.update_ddl([
    f'ALTER TABLE {SPANNER_TABLE} ADD COLUMN IF NOT EXISTS feature_v2_category_code INT64'
]).result()
print('Step 1: added feature_v2_category_code INT64.')

# Step 2: backfill from the old column with CASE (partitioned DML, server-side).
recipe_start = time.perf_counter()
recipe_rows = database.execute_partitioned_dml(f"""
    UPDATE {SPANNER_TABLE}
    SET feature_v2_category_code = CASE feature_v2_category
        WHEN 'low' THEN 0 WHEN 'medium' THEN 1 WHEN 'high' THEN 2 ELSE NULL END
    WHERE feature_v2_category IS NOT NULL
""")
recipe_ms = (time.perf_counter() - recipe_start) * 1000
print(f'Step 2: backfilled {recipe_rows:,} rows in {recipe_ms:.0f} ms via CAST/CASE.')

# Step 3: verify cut-over correctness.
with database.snapshot() as snapshot:
    check = list(snapshot.execute_sql(f"""
        SELECT feature_v2_category, feature_v2_category_code, COUNT(*) AS n
        FROM {SPANNER_TABLE} WHERE feature_v2_category IS NOT NULL
        GROUP BY 1, 2 ORDER BY 2
    """))
print('Step 3: string → code mapping:')
for cat, code_val, n in check:
    print(f'  {cat:<8s} → {code_val}  ({n:,} rows)')

# Step 4: drop the throwaway column (in a real migration you would drop the OLD
# column after readers cut over; here the new code column is the demo artifact).
drop_column('feature_v2_category_code')
print('Step 4: dropped feature_v2_category_code (demo cleanup).')
```

## 3d. DDL Timing — Median and p95

A single `ddl_ms` number is noisy. Run an `ADD COLUMN` + `DROP COLUMN` round-trip a few times and report the **median** and **p95** so the timing is meaningful. We time the backfill (partitioned DML) separately, since it scales with row count while DDL is largely fixed-cost.

```python
import statistics

def pctl(values, p):
    s = sorted(values)
    k = max(0, min(len(s) - 1, int(round((p / 100) * (len(s) - 1)))))
    return s[k]

add_times, drop_times = [], []
TRIALS = 5
for n in range(TRIALS):
    col = f'feature_timing_probe_{n}'
    t0 = time.perf_counter()
    database.update_ddl([f'ALTER TABLE {SPANNER_TABLE} ADD COLUMN IF NOT EXISTS {col} FLOAT64']).result()
    add_times.append((time.perf_counter() - t0) * 1000)
    t0 = time.perf_counter()
    drop_column(col)
    drop_times.append((time.perf_counter() - t0) * 1000)

print(f'ADD COLUMN  ({TRIALS} trials): median {statistics.median(add_times):.0f} ms, '
      f'p95 {pctl(add_times, 95):.0f} ms')
print(f'DROP COLUMN ({TRIALS} trials): median {statistics.median(drop_times):.0f} ms, '
      f'p95 {pctl(drop_times, 95):.0f} ms')
print(f'\nDDL is roughly fixed-cost (metadata operation). The earlier full-table backfill')
print(f'took {backfill_ms:.0f} ms — partitioned-DML time scales with rows touched, not schema size.')
```

## 5. Schema Version Tracking

For production feature stores, track schema changes in a metadata table. This lets the serving app know which version of the schema it should expect.

```python
# Create schema version table
version_ddl = """CREATE TABLE IF NOT EXISTS schema_versions (
    version INT64 NOT NULL,
    table_name STRING(MAX) NOT NULL,
    change_description STRING(MAX),
    columns_added ARRAY<STRING(MAX)>,
    columns_dropped ARRAY<STRING(MAX)>,
    applied_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
) PRIMARY KEY (table_name, version)"""

try:
    operation = database.update_ddl([version_ddl])
    operation.result()
    print('Created schema_versions table.')
except Exception as e:
    if 'already exists' in str(e):
        print('schema_versions table already exists.')
    else:
        print(f'Error: {e}')
```

```python
# Query schema versions
with database.snapshot() as snapshot:
    versions = list(snapshot.execute_sql("""
        SELECT version, table_name, change_description, applied_at
        FROM schema_versions
        ORDER BY table_name, version
    """))

print('Schema version history:')
for v, table, desc, ts in versions:
    print(f'  v{v} ({table}): {desc} [{ts}]')
```

### Native DDL History — `list_database_operations`

The `schema_versions` table above is a hand-rolled audit log. Spanner *also* records every DDL operation natively. `database.list_database_operations()` surfaces the long-running operations the admin API tracked — filter for `UpdateDatabaseDdl` to see the actual `ALTER`/`CREATE` statements with their commit timestamps. This complements (doesn't replace) the metadata table: the native log is automatic and authoritative, while your table carries business context like change descriptions.

```python
# Filter for UpdateDatabaseDdl operations. The metadata @type filter is the
# documented way to select DDL operations from the long-running-ops list.
ddl_filter = (
    'metadata.@type:type.googleapis.com/'
    'google.spanner.admin.database.v1.UpdateDatabaseDdlMetadata'
)

ops = list(database.list_database_operations(filter_=ddl_filter))
print(f'Spanner recorded {len(ops)} DDL operation(s) natively. Most recent:')

# list_database_operations returns api_core Operation objects whose .metadata is
# already deserialized to an UpdateDatabaseDdlMetadata (statements + commit_timestamps).
recent = []
for op in ops:
    meta = op.metadata
    if meta is None:
        continue
    timestamps = list(getattr(meta, 'commit_timestamps', []) or [])
    ts = timestamps[-1] if timestamps else None
    for stmt in (getattr(meta, 'statements', []) or []):
        recent.append((ts, stmt))

# Sort newest first by commit timestamp (None last) and show the latest handful.
recent.sort(key=lambda x: (x[0] is not None, x[0]), reverse=True)
for ts, stmt in recent[:8]:
    when = ts.isoformat() if ts is not None else 'in-progress'
    print(f'  [{when}] {stmt[:90]}')

print(f'\nNative history is automatic and authoritative; the schema_versions table adds')
print(f'human context (descriptions, owning team) on top of it.')
```

## 6. Runtime Schema Discovery

A serving app can query `INFORMATION_SCHEMA` at startup to discover which columns are available, without hardcoding column names.

```python
def discover_feature_columns(table_name, prefix='feature_'):
    """Discover all feature columns at runtime from INFORMATION_SCHEMA."""
    with database.snapshot() as snapshot:
        columns = list(snapshot.execute_sql("""
            SELECT COLUMN_NAME, SPANNER_TYPE
            FROM INFORMATION_SCHEMA.COLUMNS
            WHERE TABLE_NAME = @table AND COLUMN_NAME LIKE @prefix
            ORDER BY ORDINAL_POSITION
        """, params={'table': table_name, 'prefix': f'{prefix}%'},
             param_types={'table': spanner.param_types.STRING, 'prefix': spanner.param_types.STRING}))
    return columns

features = discover_feature_columns(SPANNER_TABLE)
print(f'Discovered {len(features)} feature columns at runtime:')

type_groups = {}
for name, stype in features:
    type_groups.setdefault(stype, []).append(name)

for stype, cols in sorted(type_groups.items(), key=lambda x: -len(x[1])):
    print(f'  {stype:<25s} {len(cols):>3} columns (e.g., {cols[0]})')

print(f'\nThe serving app uses this to build SELECT statements dynamically —')
print(f'no hardcoded column list that breaks when the schema changes.')
```

### Forward-Compatible Reader

Runtime discovery is only half the story. A serving reader must also **tolerate a column that isn't there yet** during a rolling schema change — exactly the Valkey read-with-optional pattern, but type-safe. Two complementary tactics:

1. **Build the SELECT list from discovered columns** so you never name a column that doesn't exist.
2. **Try/except around `execute_sql`** as a backstop, falling back to a default when an optional feature is genuinely absent.

This lets the model roll out *before* the backfill finishes: the reader serves a default until the feature appears, then picks it up with no code change.

```python
def read_features_forward_compatible(group, entity_id, wanted, defaults=None):
    """Read `wanted` features, tolerating any that don't exist yet.

    Tactic 1: intersect the request with columns actually present (from
    INFORMATION_SCHEMA), so the generated SELECT never references a missing column.
    Tactic 2: if a wanted column is absent, serve its default instead of failing.
    """
    defaults = defaults or {}
    present = {name for name, _ in discover_feature_columns(SPANNER_TABLE, prefix='feature_')}
    present |= {'entity_group', 'entity_id'}
    available = [c for c in wanted if c in present]
    missing = [c for c in wanted if c not in present]

    select_list = ', '.join(available)
    with database.snapshot(multi_use=True) as snapshot:
        row = list(snapshot.execute_sql(
            f"SELECT {select_list} FROM {SPANNER_TABLE} "
            f"WHERE entity_group = @g AND entity_id = @e",
            params={'g': group, 'e': entity_id},
            param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING},
        ))
    out = dict(zip(available, row[0])) if row else {}
    for c in missing:
        out[c] = defaults.get(c)  # serve default for not-yet-added columns
    return out, missing

# Ask for a real column AND one that does not exist yet (simulating mid-rollout).
wanted = ['feature_float_1', 'feature_int_1', 'feature_not_yet_added']
result, missing = read_features_forward_compatible(
    'A', '000001', wanted, defaults={'feature_not_yet_added': 0.0})

print('Forward-compatible read:')
for col in wanted:
    tag = '  (default — column absent)' if col in missing else ''
    print(f'  {col:<25s} = {result[col]}{tag}')

# Backstop demonstration: a raw query for a missing column fails; we catch it.
try:
    with database.snapshot(multi_use=True) as snapshot:
        list(snapshot.execute_sql(
            f"SELECT feature_not_yet_added FROM {SPANNER_TABLE} LIMIT 1"))
    print('\nUnexpected: query for missing column succeeded.')
except Exception as e:
    print(f'\nRaw SELECT of a missing column raises {type(e).__name__} — the discovery-driven')
    print('SELECT list above avoids this entirely; try/except is the backstop.')
```

## Schema-Change Compatibility Matrix

Not every schema change is online or even allowed. This is the practical map of what Spanner permits, what it validates (and can therefore reject), and what forces the add/backfill/drop recipe from section 3c.

| Change | Allowed? | Notes |
|---|---|---|
| `ADD COLUMN` (nullable) | Yes, online | New column reads `NULL` for existing rows until backfilled |
| `ADD COLUMN ... DEFAULT (expr)` | Yes, online | Existing rows read the default immediately — no backfill (§3a) |
| `ADD COLUMN ... AS (...) STORED` | Yes (backfilled) | Generated column; materialized on a background backfill (§3b) |
| `DROP COLUMN` | Yes, online | Data is removed; not recoverable without backup |
| `STRING(50)` → `STRING(MAX)` | Yes, online | Widening length is metadata-only |
| `STRING(MAX)` → `STRING(50)` | Validated | Spanner scans to confirm all values fit; **fails** if any value is too long |
| `STRING` → `BYTES` (and `BYTES` → `STRING`) | Yes, validated | Allowed because the encodings are compatible; validated for UTF-8 |
| Add `NOT NULL` to existing column | Validated | **Fails** if any existing row holds `NULL`; backfill first |
| `ARRAY<T1>` → `ARRAY<T2>` (different element type) | Forbidden | Array element type cannot change |
| `STRING` → `INT64` (incompatible types) | Forbidden | Use add new column + backfill with `CAST` + drop old (§3c) |

**Rule of thumb:** *widening* and *encoding-compatible* changes are online; *narrowing* and *constraint-tightening* changes are validated (and may fail on bad data); *incompatible type* changes require the recipe.

## 7. Comparison to Bigtable Schema Management

| Aspect | Spanner | Bigtable |
|---|---|---|
| **Schema model** | Strongly typed, DDL-defined | Schema-free (any column qualifier) |
| **Add column** | `ALTER TABLE ADD COLUMN` (online) | Just write to a new qualifier (no DDL) |
| **Drop column** | `ALTER TABLE DROP COLUMN` (online) | Stop writing to qualifier (no cleanup) |
| **Type enforcement** | At write time (database rejects wrong types) | At read time (app must decode bytes) |
| **Schema discovery** | `INFORMATION_SCHEMA` (built-in) | Custom `#schema` metadata row |
| **Default values** | `DEFAULT` clause; existing rows read it instantly | N/A (app supplies defaults) |
| **Generated columns** | `AS (...) STORED` — DB keeps derived features consistent | **No analog** (app computes derived features) |
| **Type change** | Add new column + backfill `CAST` + drop old | Just write new bytes; reader interprets |
| **Version tracking** | Metadata table **+** native `list_database_operations` | Custom `#registry` row |
| **Backfill** | Partitioned DML (server-side) | Batch mutations (client-side) |

**Trade-off:** Bigtable's schema freedom means you can add features instantly without DDL. Spanner's strong typing means you catch errors at write time instead of discovering them at serving time — and features like generated columns and validated type changes are simply not available in a schema-free store.

## Summary

| Operation | Method | Downtime | Section |
|---|---|---|---|
| Add column | `ALTER TABLE ADD COLUMN` | None (online) | 2 |
| Backfill values | Partitioned DML (no commit-ts column) | None | 3 |
| Audit timestamp | `PENDING_COMMIT_TIMESTAMP()` in regular DML txn | None | 3 |
| Default-value column | `ADD COLUMN ... DEFAULT (expr)` | None (instant for existing rows) | 3a |
| Generated column | `ADD COLUMN ... AS (...) STORED` | Backfilled | 3b |
| Type change | Add new col → backfill `CAST` → drop old | None | 3c |
| DDL timing | median / p95 over repeated trials | N/A | 3d |
| Drop column | `ALTER TABLE DROP COLUMN` | None (online) | 4 |
| Version tracking | Metadata table + `list_database_operations` | N/A | 5 |
| Schema discovery | `INFORMATION_SCHEMA` + forward-compatible reader | N/A | 6 |

**Key takeaways**
- Online DDL is real: reads and writes kept flowing *during* the schema change (section 2 proof).
- Partitioned DML cannot set a commit-timestamp column — backfill values there, stamp `updated_at` in a regular transaction.
- Generated `STORED` columns have no Bigtable/Valkey analog: the database keeps derived features consistent for you.
- Incompatible type changes are forbidden in place — the add/backfill/drop recipe is the canonical path.
- Forward-compatible readers (discovery-driven SELECT + try/except fallback) let models roll out before backfills finish.
- Every demo column added here is dropped again, so the table schema is left exactly as later notebooks expect.
