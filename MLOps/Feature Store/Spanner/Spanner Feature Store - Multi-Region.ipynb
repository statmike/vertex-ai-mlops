# Spanner Feature Store — Multi-Region and High Availability

Spanner's multi-regional instances provide automatic failover with strong consistency — the [99.999% SLA](https://cloud.google.com/spanner/sla). This notebook covers the architecture, configuration, and latency implications of multi-regional deployment.

**What you'll learn:**
- Multi-regional instance configurations (`nam3`, `nam-eur-asia1`, etc.)
- Read-write vs read-only replicas and witness replicas
- Automatic failover on zone or region failure
- Stale reads served locally (no cross-region round-trip)
- Strong reads and the leader round-trip penalty
- Cost implications of multi-regional deployment (~3× base)

**Key difference from Bigtable:** Bigtable multi-cluster replication is eventually consistent and requires manual app profile configuration for routing. Spanner multi-region is strongly consistent with automatic failover — no app profiles, no routing decisions.

**Note:** This notebook is primarily conceptual — creating a multi-regional instance requires significantly more compute resources and cost. The concepts and configurations are demonstrated with reference to the existing regional instance.

**Prerequisites:** Run [Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) first.

## 1. Instance Configurations

Spanner offers regional and multi-regional configurations:

### Regional (Current Setup)

```
regional-us-central1
└── 3 zones within us-central1
    ├── Zone A: read-write replica
    ├── Zone B: read-write replica
    └── Zone C: read-write replica
SLA: 99.99%
```

### Multi-Regional Options

| Config | Regions | Replica Placement | SLA |
|---|---|---|---|
| `nam3` | N. Virginia (`us-east4`), S. Carolina (`us-east1`), Iowa (`us-central1`) | 2 R/W regions (N. Virginia = default leader) + 1 witness (Iowa) | 99.999% |
| `nam6` | Iowa (`us-central1`), S. Carolina (`us-east1`), Oregon (`us-west1`), Los Angeles (`us-west2`), Oklahoma (`us-central2`) | 2 R/W (Iowa = default leader, S. Carolina) + 2 R/O (Oregon, L.A.) + 1 witness (Oklahoma) | 99.999% |
| `nam-eur-asia1` | Iowa, Oklahoma, Belgium, Netherlands, Taiwan, Singapore | 2 R/W (N. America) + R/O continents + witnesses | 99.999% |
| `eur3` | Belgium (`europe-west1`), Netherlands (`europe-west4`), Finland (`europe-north1`) | 2 R/W + 1 witness | 99.999% |

> The exact replica composition above is **read from the API** in the next code cell so the table can never drift from reality. Source: [instance configurations](https://cloud.google.com/spanner/docs/instance-configurations).

### Replica Types

| Type | Purpose | Serves Reads | Serves Writes |
|---|---|---|---|
| **Read-write** | Full replica, participates in write (Paxos) quorum, can become leader | Yes | Yes |
| **Read-only** | Serves reads, does **not** vote in the write quorum, never becomes leader | Yes | No |
| **Witness** | Votes in the write quorum only — stores no row data | **No** | Vote only |

> **Common misconception:** a witness replica does *not* serve reads. It exists purely to make a quorum reachable (an odd number of voters) without paying for a third full data copy. In `nam3`, that means reads are only ever served by the two R/W regions.


````python
# Build the configuration/replica table FROM the API so it is self-correcting.
# (Hand-typed replica lists drift out of date — let the API be the source of truth.)
configs = list(spanner_client.list_instance_configs())
print(f'Available Spanner configurations ({len(configs)} total)\n')

# A config is multi-region when it has replicas in more than one distinct location.
# Match the well-known multi-region families AND fall back to a generic detector so
# we never silently miss asia / nam-eur-asia / custom configs.
MR_PREFIXES = ('nam', 'eur', 'asia', 'aus', 'india')  # known multi-region families

for config in sorted(configs, key=lambda c: c.name):
    short = config.name.split('/')[-1]
    replicas = list(getattr(config, 'replicas', []) or [])
    locations = {r.location for r in replicas}
    is_multi_region = len(locations) > 1 or short.startswith(MR_PREFIXES)
    if not is_multi_region:
        continue
    print(f'{short}  ({config.display_name})')
    if not replicas:
        print('    (replica detail not returned by this API version)')
    for r in replicas:
        # r.type_ is a ReplicaType enum (READ_WRITE / READ_ONLY / WITNESS)
        leader = ' [default leader]' if getattr(r, 'default_leader_location', False) else ''
        print(f'    - {r.location:<16s} {r.type_.name:<11s}{leader}')
    print()
````

### Which Replica Serves Which Read?

The read type you request and the replica type that receives it together decide whether a request pays a cross-region round-trip to the leader. This is the single most important table for reasoning about multi-region latency:

| Read type | Read-write replica | Read-only replica | Witness |
|---|---|---|---|
| **Strong** (`strong=True`) | Serves locally **if** it already has the latest data; otherwise waits for / round-trips the leader | Same: may need a leader round-trip to confirm it is current | **Never** — witnesses hold no data |
| **Bounded staleness** (`max_staleness=…`) | Serves locally if its data is within the bound; only contacts leader when too stale | Serves locally if within the bound | **Never** |
| **Exact staleness** (`exact_staleness=…`) | Serves locally — picks a read timestamp in the past, no leader contact | Serves locally — no leader contact | **Never** |

Key takeaways for feature serving:
- **Strong reads** guarantee you see the latest committed write, but a non-leader replica may have to confirm currency with the leader → up to one cross-region RTT.
- **Stale reads** (bounded or exact) can almost always be answered from the nearest R/W or R/O replica with **zero** leader contact — this is where multi-region latency wins come from.
- **Witnesses serve no reads at all.** Routing a read "to the witness region" is impossible; the request lands on the nearest data-bearing replica instead.


### Pinning Reads to a Region with `DirectedReadOptions`

By default Spanner routes a read to whatever replica can serve it fastest. When you want to **force** reads to a specific region — the Spanner analog of a Bigtable app-profile's single-cluster routing — you attach `DirectedReadOptions`.

This is most useful for:
- Keeping a regional analytics or batch-serving job's reads inside one region (predictable latency, no surprise cross-region traffic).
- Co-locating feature reads with compute in a particular region.

`DirectedReadOptions` only takes effect on a **multi-region** instance; on our regional instance there is exactly one region, so it is a no-op. We construct the object below to show the exact API.

````python
# DirectedReadOptions: pin reads to chosen replicas/regions (multi-region only).
from google.cloud.spanner_v1 import DirectedReadOptions

# Prefer the us-east1 (South Carolina) R/W replica, then fall back to us-east4.
# auto_failover_disabled=False lets Spanner fall back to other replicas if the
# listed ones are unavailable (set True to fail the read instead).
directed_read_options = DirectedReadOptions(
    include_replicas=DirectedReadOptions.IncludeReplicas(
        replica_selections=[
            DirectedReadOptions.ReplicaSelection(
                location='us-east1',
                type_=DirectedReadOptions.ReplicaSelection.Type.READ_WRITE,
            ),
            DirectedReadOptions.ReplicaSelection(
                location='us-east4',
                type_=DirectedReadOptions.ReplicaSelection.Type.READ_WRITE,
            ),
        ],
        auto_failover_disabled=False,
    )
)

print('Constructed DirectedReadOptions:')
print(directed_read_options)

# Usage on a multi-region instance (no-op on our regional instance):
#   with database.snapshot() as snapshot:
#       rows = snapshot.execute_sql(
#           query,
#           directed_read_options=directed_read_options,
#       )
# Note: directed reads are rejected on read-write transactions — they are for
# read-only snapshots / single-use reads only.
print('\nOn a regional instance this has no effect (one region only).')
````

## 2. Multi-Regional Architecture for Feature Stores

For a feature store, multi-regional Spanner provides (using the **`nam3`** configuration):

```
nam3 configuration (default leader: N. Virginia):

N. Virginia (us-east4)       South Carolina (us-east1)     Iowa (us-central1)
├─ Read-Write Replica        ├─ Read-Write Replica         └─ Witness
├─ Default leader            ├─ Serves reads (strong)         (quorum vote only,
├─ Serves reads (strong)     └─ Serves reads (stale)           stores NO data,
└─ Serves reads (stale)                                        serves NO reads)

Writes: go to the leader (N. Virginia), replicated to a quorum before commit.
        The Iowa witness votes in the quorum but holds no data.
Strong reads: a non-leader replica may round-trip the leader (cross-region latency).
Stale reads: served from the nearest R/W replica (no cross-region needed).
```

> **Correction note:** the two read-write regions in `nam3` are **N. Virginia (`us-east4`)** and **South Carolina (`us-east1`)**, with **Iowa (`us-central1`)** as the witness. Oklahoma (`us-central2`) is **not** part of `nam3` — it is the witness region in `nam6`.

### Why Stale Reads Matter for Multi-Region

Strong reads may require a round-trip to the leader region. Stale reads (even just 10–15 seconds) can be served from the local replica, eliminating cross-region latency.

For feature serving, this is almost always acceptable — ML features that were current 10 seconds ago are fine for most models.

### Illustrative Cross-Region Round-Trip Latencies

The "cross-region penalty" on a strong read is, concretely, the network round-trip to the leader region. The numbers below are **illustrative** order-of-magnitude RTTs between US regions — actual values vary with time of day, path, and Google's backbone, so treat them as ballpark, not SLA.

| Region pair | Approx. one-way | Approx. round-trip (RTT) |
|---|---|---|
| `us-central1` (Iowa) ↔ `us-east4` (N. Virginia) | ~12 ms | **~25 ms** |
| `us-central1` (Iowa) ↔ `us-east1` (S. Carolina) | ~13 ms | **~26 ms** |
| `us-east1` (S. Carolina) ↔ `us-east4` (N. Virginia) | ~6 ms | **~12 ms** |
| `us-central1` (Iowa) ↔ `us-west1` (Oregon) | ~20 ms | **~40 ms** |
| `us-east4` (N. Virginia) ↔ `us-west2` (L.A.) | ~30 ms | **~60 ms** |

This is what grounds the "stale reads save ~20–40 ms" claim further down: on `nam3`, a strong read landing on the S. Carolina replica that must confirm currency with the N. Virginia leader pays roughly **one ~12 ms RTT**, while a read landing in a far region pays much more. A stale read avoids that round-trip entirely by serving from the local replica.

````python
# Demonstrate stale read latency advantage on our regional instance
database = instance.database(SPANNER_DATABASE)
SPANNER_TABLE = 'dense_features'

n_reads = 30

query = f"SELECT feature_float_1 FROM {SPANNER_TABLE} WHERE entity_group = 'A' AND entity_id = '000001'"

strong_lats = []
for _ in range(n_reads):
    start = time.perf_counter()
    with database.snapshot(multi_use=True) as snapshot:
        list(snapshot.execute_sql(query))
    strong_lats.append((time.perf_counter() - start) * 1000)

stale_lats = []
for _ in range(n_reads):
    start = time.perf_counter()
    with database.snapshot(exact_staleness=datetime.timedelta(seconds=15)) as snapshot:
        list(snapshot.execute_sql(query))
    stale_lats.append((time.perf_counter() - start) * 1000)

print(f'Read latency comparison ({n_reads} reads):\n')
print(f'{"Mode":<20s} {"Median":>10s} {"p95":>10s} {"Mean":>10s}')
for name, lats in [('Strong', strong_lats), ('Stale (15s)', stale_lats)]:
    s = sorted(lats)
    print(f'{name:<20s} {s[n_reads//2]:>8.1f}ms {s[int(n_reads*0.95)]:>8.1f}ms {sum(lats)/len(lats):>8.1f}ms')

print(f'\nOn a regional instance, the difference is modest.')
print(f'On multi-regional (nam3), stale reads avoid a leader round-trip,')
print(f'saving roughly the cross-region RTT (~12 ms within the US East,')
print(f'tens of ms to far regions — see the illustrative RTT table above).')
````

## 3. Proof: Read-Your-Write Is Immediate (Zero Replication Lag)

Spanner's headline claim is **strong consistency with zero data loss**: the moment a write commits, a strong read sees it — there is no eventual-consistency window to wait out (contrast the Bigtable Replication notebook, which has to *measure* how long an async-replicated write takes to appear in the other cluster).

We prove it on the existing **regional** instance by writing a scratch row in a transaction, then immediately issuing a fresh **strong** snapshot read. The value is visible with zero lag. We clean the scratch row up afterward. (The same guarantee holds on multi-region — there a strong read additionally confirms currency with the leader, which is the cross-region cost discussed above.)

````python
# Read-your-write proof: commit a scratch row, then strong-read it immediately.
# entity_group is STRING(1) and entity_id is STRING(6); use 'Z' + a non-numeric id
# so we never collide with real data (000001-005000). updated_at is NOT NULL.
SCRATCH_GROUP = 'Z'
SCRATCH_ID = 'RYWPRF'
SCRATCH_TABLE = 'dense_features'
sentinel = time.time()  # unique marker so we know we read THIS write

# 1. Write in a read-write transaction (commit returns a commit timestamp).
def _write(txn):
    txn.insert_or_update(
        SCRATCH_TABLE,
        columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
        values=[(SCRATCH_GROUP, SCRATCH_ID, sentinel, spanner.COMMIT_TIMESTAMP)],
    )

commit_ts = database.run_in_transaction(_write)
print(f'Wrote scratch row, commit timestamp: {commit_ts}')

# 2. Immediately read it back with a STRONG snapshot (default = strong).
read_query = (
    f"SELECT feature_float_1 FROM {SCRATCH_TABLE} "
    f"WHERE entity_group = @g AND entity_id = @e"
)
with database.snapshot() as snapshot:
    rows = list(snapshot.execute_sql(
        read_query,
        params={'g': SCRATCH_GROUP, 'e': SCRATCH_ID},
        param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING},
    ))

read_value = rows[0][0] if rows else None
print(f'Strong read returned: {read_value}')
print(f'Matches the value we just wrote: {read_value == sentinel}')
print('=> zero replication lag: the write is visible to a strong read immediately.')

# 3. Clean up the scratch row.
def _delete(txn):
    txn.delete(
        SCRATCH_TABLE,
        keyset=spanner.KeySet(keys=[(SCRATCH_GROUP, SCRATCH_ID)]),
    )

database.run_in_transaction(_delete)
print('Scratch row deleted.')
````

## 4. Automatic Failover

Spanner handles failover automatically — no manual intervention, no app profile reconfiguration, no data loss.

| Failure | Bigtable | Spanner |
|---|---|---|
| Zone failure | Must have multi-cluster + app profiles configured | Automatic (regional: 3 zones) |
| Region failure | Multi-cluster replication + manual failover | Automatic (multi-regional config) |
| Data loss risk | Eventual consistency between clusters | Zero (synchronous quorum writes) |
| Client changes | Update app profile routing | None |
| Recovery time | Seconds (with proper app profiles) | Transparent (no recovery needed) |

### How Writes Work in Multi-Region

1. Client sends write to any Spanner frontend
2. Frontend routes to the **leader** (in one of the R/W regions)
3. Leader proposes write to quorum (majority of voting replicas)
4. Quorum acknowledges → write committed
5. Non-quorum replicas catch up asynchronously

If the leader fails mid-write, a new leader is elected from the remaining R/W replicas. The client retries transparently.

## 5. Cost Considerations

Multi-regional Spanner is expensive — the data is stored across multiple regions with synchronous replication.

| Config | PU Cost Multiplier | Storage Multiplier | Use Case |
|---|---|---|---|
| Regional | 1× | 1× | Development, cost-sensitive workloads |
| Multi-regional (nam3) | ~3× | ~3× | Production with HA requirements |
| Multi-continental | ~3–5× | ~3–5× | Global serving with geo-local reads |

**Feature store recommendation:**
- **Development/testing**: Regional (100 PU) — what we use in this series
- **Production (single region)**: Regional with sufficient PUs for throughput
- **Production (HA required)**: Multi-regional (nam3 or similar) — 99.999% SLA
- **Global serving**: Multi-continental — stale reads from nearest region

### What the Extra "Nine" Actually Buys: Downtime Math

The ~3× cost jump from regional (99.99%) to multi-region (99.999%) buys exactly one more "nine" of availability. Translated into allowed annual downtime:

| SLA | Allowed downtime / year | Allowed / month | Allowed / day |
|---|---|---|---|
| 99.99% (regional) | **~52.6 min** | ~4.4 min | ~8.6 s |
| 99.999% (multi-region) | **~5.26 min** | ~26 s | ~0.86 s |

Math: `(1 - 0.9999) × 365.25 × 24 × 60 ≈ 52.6 min`; one more nine divides that by 10 → `≈ 5.26 min`. So the multi-region premium removes ~47 minutes of *permitted* outage per year and survives a full **region** failure (not just a zone). Whether that is worth ~3× depends on the cost of an hour of stale or unavailable features to your serving path.

## 6. Comparison to Bigtable Replication

| Aspect | Spanner Multi-Region | Bigtable Multi-Cluster |
|---|---|---|
| **Consistency** | Strong (synchronous quorum) | Eventual (async replication) |
| **Failover** | Automatic, transparent | Manual app profile switch (or multi-cluster routing) |
| **Configuration** | Choose instance config (e.g., `nam3`) | Create clusters + app profiles |
| **Client changes** | None | Update app profile routing |
| **Data loss risk** | Zero | Possible (eventual consistency window) |
| **SLA** | 99.999% | 99.999% (with proper setup) |
| **Cost** | ~3× base | ~2× (per-node for each cluster) |
| **Latency trade-off** | Stale reads avoid cross-region | App profile routing to nearest cluster |

**When to choose Spanner multi-region:**
- Strong consistency required across regions
- Zero data loss on region failure
- No operational overhead for failover management

**When to choose Bigtable multi-cluster:**
- Eventual consistency acceptable
- Lower cost for multi-region serving
- Need fine-grained control over routing (app profiles)
