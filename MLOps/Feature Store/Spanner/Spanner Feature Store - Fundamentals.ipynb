# Spanner Feature Store — Fundamentals

End-to-end flow from BigQuery to Spanner to serving. After this notebook you'll have all 130K entities loaded in Spanner and a working read path with latency baselines.

**What you'll learn:**
- Load all 130K entities from BigQuery into Spanner using batch mutations
- Composite primary key: `(entity_group, entity_id)` — no row key concatenation needed
- Point reads by primary key using SQL and the read API
- Range scans, batch reads, and entity group queries
- `INFORMATION_SCHEMA` as the self-describing schema registry
- Query plan visualization with `EXPLAIN`
- Latency baseline: strong reads vs stale reads
- Comparison to Bigtable: SQL vs row key lookups, native types vs byte decoding

**Prerequisites:** Run [Spanner Feature Store — Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) first.

## 1. Full Data Load — BigQuery to Spanner

The Environment notebook loaded a sample (entity group A). Here we load all 130K entities from BigQuery into Spanner using batch mutations.

In a production feature store, you'd use [continuous queries](https://cloud.google.com/bigquery/docs/continuous-queries-introduction) for ongoing sync (covered in Notebook 2). This batch load establishes the baseline dataset.

### Strategy

1. Read from BigQuery in chunks (one entity group at a time)
2. Convert each chunk to Spanner-compatible types
3. Insert using `database.batch()` with 500-row batches
4. Use `INSERT_OR_UPDATE` to make the load idempotent

```python
# Check current row count
with database.snapshot() as snapshot:
    current_count = list(snapshot.execute_sql(f'SELECT COUNT(*) FROM {SPANNER_TABLE}'))[0][0]

print(f'Current rows in Spanner: {current_count:,}')
print(f'Target: 130,000 (26 groups × 5,000 entities)')

if current_count >= 130000:
    print('\nAll data already loaded — skipping.')
    SKIP_LOAD = True
else:
    print(f'\nNeed to load {130000 - current_count:,} more rows.')
    SKIP_LOAD = False
```

```python
# Columns to load (a practical subset — the most common feature types)
LOAD_COLUMNS = [
    'entity_group', 'entity_id',
    'feature_int_1', 'feature_int_2', 'feature_int_3', 'feature_int_4', 'feature_int_5',
    'feature_float_1', 'feature_float_2', 'feature_float_3', 'feature_float_4', 'feature_float_5',
    'feature_numeric_1', 'feature_numeric_2', 'feature_numeric_3',
    'feature_string_1', 'feature_string_2', 'feature_string_3', 'feature_string_4', 'feature_string_5',
    'feature_bool_1', 'feature_bool_2', 'feature_bool_3',
    'feature_date_1', 'feature_date_2', 'feature_date_3',
    'label',
]

# Add the bulk float columns
LOAD_COLUMNS += [f'feature_float_{i}' for i in range(6, 181)]

SPANNER_COLUMNS = LOAD_COLUMNS + ['updated_at']

print(f'Loading {len(LOAD_COLUMNS)} columns from BigQuery')
print(f'Writing {len(SPANNER_COLUMNS)} columns to Spanner (+ updated_at)')
```

```python
def convert_row(row):
    """Convert a pandas row to Spanner-compatible values."""
    values = []
    for col in LOAD_COLUMNS:
        val = row[col]
        if pd.isna(val):
            values.append(None)
        elif isinstance(val, (np.integer,)):
            values.append(int(val))
        elif isinstance(val, (np.floating,)):
            values.append(float(val))
        elif isinstance(val, (np.bool_,)):
            values.append(bool(val))
        elif isinstance(val, datetime.date) and not isinstance(val, datetime.datetime):
            values.append(val)
        else:
            values.append(val)
    values.append(spanner.COMMIT_TIMESTAMP)  # updated_at
    return values

print('Row converter ready.')
```

```python
if not SKIP_LOAD:
    cols_sql = ', '.join(LOAD_COLUMNS)
    groups = [chr(c) for c in range(65, 91)]  # A-Z
    # Spanner caps a commit at 80,000 mutations, counted as rows x columns-affected
    # (plus any index entries). This wide table has many columns, so size the batch
    # to stay safely under the limit: rows_per_batch x len(SPANNER_COLUMNS) < 80,000.
    MUTATION_LIMIT = 80_000
    batch_size = max(1, int(MUTATION_LIMIT * 0.9) // len(SPANNER_COLUMNS))  # 10% safety margin
    print(f'Batch size: {batch_size} rows x {len(SPANNER_COLUMNS)} cols '
          f'= {batch_size * len(SPANNER_COLUMNS):,} mutations/commit (limit 80,000)')
    total_loaded = 0
    overall_start = time.perf_counter()

    for group in groups:
        # Check if this group already exists
        with database.snapshot() as snapshot:
            group_count = list(snapshot.execute_sql(
                f"SELECT COUNT(*) FROM {SPANNER_TABLE} WHERE entity_group = @g",
                params={'g': group}, param_types={'g': spanner.param_types.STRING}
            ))[0][0]

        if group_count >= 5000:
            total_loaded += group_count
            continue

        group_start = time.perf_counter()

        df = bq.query(f'''
            SELECT {cols_sql}
            FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
            WHERE entity_group = '{group}'
            ORDER BY entity_id
        ''').to_dataframe()

        rows = [convert_row(row) for _, row in df.iterrows()]

        for i in range(0, len(rows), batch_size):
            batch = rows[i:i + batch_size]
            with database.batch() as batcher:
                batcher.insert_or_update(
                    table=SPANNER_TABLE,
                    columns=SPANNER_COLUMNS,
                    values=batch,
                )

        elapsed = time.perf_counter() - group_start
        total_loaded += len(rows)
        print(f'  Group {group}: {len(rows):,} rows in {elapsed:.1f}s ({total_loaded:,} total)')

    overall_elapsed = time.perf_counter() - overall_start
    print(f'\nLoad complete: {total_loaded:,} rows in {overall_elapsed:.0f}s')
else:
    print('Load skipped — all data present.')
```

### Verify Full Load

```python
with database.snapshot(multi_use=True) as snapshot:
    total = list(snapshot.execute_sql(f'SELECT COUNT(*) FROM {SPANNER_TABLE}'))[0][0]

    groups = list(snapshot.execute_sql(f'''
        SELECT entity_group, COUNT(*) AS n
        FROM {SPANNER_TABLE}
        GROUP BY entity_group
        ORDER BY entity_group
    '''))

print(f'Total rows: {total:,}')
print(f'Entity groups: {len(groups)}')
print(f'Rows per group: {groups[0][1]:,} (first), {groups[-1][1]:,} (last)')
assert total == 130000, f'Expected 130,000 rows, got {total:,}'
assert len(groups) == 26, f'Expected 26 groups, got {len(groups)}'
print('\n✔ All 130,000 rows across 26 entity groups verified.')
```

## 2. INFORMATION_SCHEMA — Self-Describing Schema

In the [Bigtable Feature Store](../Bigtable/readme.md), we store a `#schema` metadata row so clients can decode feature values without hardcoded column maps. Spanner doesn't need this — the schema is built into the database and queryable via `INFORMATION_SCHEMA`.

This is a significant operational advantage: schema changes via DDL (`ALTER TABLE`) automatically update `INFORMATION_SCHEMA`, so the serving app always sees the current schema without any custom sync logic.

```python
with database.snapshot(multi_use=True) as snapshot:
    schema = list(snapshot.execute_sql("""
        SELECT COLUMN_NAME, SPANNER_TYPE, IS_NULLABLE, ORDINAL_POSITION
        FROM INFORMATION_SCHEMA.COLUMNS
        WHERE TABLE_NAME = @table
        ORDER BY ORDINAL_POSITION
    """, params={'table': SPANNER_TABLE}, param_types={'table': spanner.param_types.STRING}))

# Build type distribution
type_counts = {}
for col in schema:
    t = col[1]
    type_counts[t] = type_counts.get(t, 0) + 1

print(f'INFORMATION_SCHEMA for {SPANNER_TABLE}: {len(schema)} columns\n')
print('Type distribution:')
for t, count in sorted(type_counts.items(), key=lambda x: -x[1]):
    print(f'  {t:<30s} {count:>3}')

print(f'\nPrimary key columns:')
with database.snapshot(multi_use=True) as snapshot:
    pk = list(snapshot.execute_sql("""
        SELECT COLUMN_NAME, COLUMN_ORDERING
        FROM INFORMATION_SCHEMA.INDEX_COLUMNS
        WHERE TABLE_NAME = @table AND INDEX_NAME = 'PRIMARY_KEY'
        ORDER BY ORDINAL_POSITION
    """, params={'table': SPANNER_TABLE}, param_types={'table': spanner.param_types.STRING}))
for col_name, ordering in pk:
    print(f'  {col_name} ({ordering})')
```

## 3. Point Reads — Single Entity by Primary Key

The most common feature store operation: look up one entity by its key. Spanner offers two ways to do this:

1. **SQL query**: `SELECT * FROM table WHERE entity_group = 'A' AND entity_id = '000001'`
2. **Read API**: `database.snapshot().read(table, columns, keyset)` — bypasses SQL parsing

Both use the primary key index. The Read API is slightly faster because it skips SQL parsing and optimization.

```python
# SQL point read
FEATURE_COLUMNS = [
    'entity_group', 'entity_id',
    'feature_int_1', 'feature_float_1', 'feature_string_1',
    'feature_bool_1', 'feature_date_1', 'label',
]

start = time.perf_counter()
with database.snapshot() as snapshot:
    result = list(snapshot.execute_sql(f"""
        SELECT {', '.join(FEATURE_COLUMNS)}
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id = '000001'
    """))
sql_ms = (time.perf_counter() - start) * 1000

print(f'SQL point read: {sql_ms:.1f} ms')
print(f'\nResult:')
for col, val in zip(FEATURE_COLUMNS, result[0]):
    print(f'  {col:<25s} {val}')
```

```python
# Read API vs SQL — statistically real comparison.
# Loop each path 50x over rotating keys and report p50 / p95.

def percentiles(xs):
    s = sorted(xs)
    p = lambda q: s[min(len(s) - 1, int(len(s) * q))]
    return p(0.50), p(0.95)

N = 50
keys = [['A', f'{(i % 5000) + 1:06d}'] for i in range(N)]

# SQL path
sql_lat = []
for g, e in keys:
    start = time.perf_counter()
    with database.snapshot(multi_use=True) as snapshot:
        list(snapshot.execute_sql(
            f"""SELECT {', '.join(FEATURE_COLUMNS)} FROM {SPANNER_TABLE}
                WHERE entity_group = @g AND entity_id = @e""",
            params={'g': g, 'e': e},
            param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING},
        ))
    sql_lat.append((time.perf_counter() - start) * 1000)

# Read API path (KeySet — bypasses the SQL parser/optimizer)
read_lat = []
for g, e in keys:
    start = time.perf_counter()
    with database.snapshot(multi_use=True) as snapshot:
        list(snapshot.read(
            table=SPANNER_TABLE,
            columns=FEATURE_COLUMNS,
            keyset=spanner.KeySet(keys=[[g, e]]),
        ))
    read_lat.append((time.perf_counter() - start) * 1000)

sql_p50, sql_p95 = percentiles(sql_lat)
read_p50, read_p95 = percentiles(read_lat)

print(f'Point read, {N} iterations each:\n')
print(f'{"":<12s} {"p50":>10s} {"p95":>10s}')
print(f'{"SQL":<12s} {sql_p50:>8.1f}ms {sql_p95:>8.1f}ms')
print(f'{"Read API":<12s} {read_p50:>8.1f}ms {read_p95:>8.1f}ms')
print(f'\nRead API p50 is {sql_p50 - read_p50:+.1f}ms vs SQL ({read_p50 / sql_p50 * 100:.0f}% of SQL latency).')
```

**Why the Read API is faster.** `snapshot.read()` takes a table, a column list, and a `KeySet`, then goes straight to the storage layer. `snapshot.execute_sql()` must first **parse** the SQL text, **analyze** types, and **plan/optimize** the query before it can fetch the same rows. For a primary-key point lookup the two read identical data, so the SQL parse + plan is pure overhead. The gap is small in absolute terms (the network round trip dominates), but at serving QPS it removes a fixed per-request CPU cost — which is why key-based serving paths often prefer the Read API. The moment a query needs a non-key filter, a join, or an aggregation, SQL is the only option.

### Write Path: Mutations vs DML

Spanner offers two ways to write data, and the loader above used the first one.

1. **Mutations API** — `batch.insert_or_update(...)`. You describe *what* the rows should look like and Spanner applies them as a buffered set of changes at commit. This is what the bulk loader uses, and it is the fastest path for blind writes (no read required).
2. **DML** — `transaction.execute_update("UPDATE ... / INSERT OR UPDATE ... SELECT ...")`. You send SQL that Spanner parses, plans, and executes inside a read-write transaction. DML can read-then-write in one statement (e.g. `UPDATE ... SET x = x + 1`) and can write rows it computes from a `SELECT`.

**The 80,000-mutation commit limit.** A single committed transaction may contain at most **80,000 mutations**, where every column value written *and every secondary-index entry it touches* counts as one mutation. With ~200 columns per row, a few hundred rows is enough to approach the ceiling — which is exactly why the loader batches **~500 rows per commit** rather than committing all 5,000 rows in a group at once. DML statements are subject to the same per-commit limit on the rows they modify.

```python
# Tiny demo: write one entity via the Mutations API, then update one column via DML.
demo_group, demo_eid = 'A', '000001'

# --- Path 1: Mutations API (blind insert_or_update) ---
with database.batch() as batcher:
    batcher.insert_or_update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_int_1', 'updated_at'],
        values=[[demo_group, demo_eid, 4242, spanner.COMMIT_TIMESTAMP]],
    )
print('Mutations API: insert_or_update committed.')

# --- Path 2: DML inside a read-write transaction ---
def _bump_via_dml(transaction):
    return transaction.execute_update(
        f"""UPDATE {SPANNER_TABLE}
            SET feature_int_2 = 9999
            WHERE entity_group = @g AND entity_id = @e""",
        params={'g': demo_group, 'e': demo_eid},
        param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING},
    )

rows_updated = database.run_in_transaction(_bump_via_dml)
print(f'DML: UPDATE modified {rows_updated} row(s).')

# Verify both writes landed
with database.snapshot() as snapshot:
    row = list(snapshot.execute_sql(
        f"""SELECT feature_int_1, feature_int_2 FROM {SPANNER_TABLE}
            WHERE entity_group = @g AND entity_id = @e""",
        params={'g': demo_group, 'e': demo_eid},
        param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING},
    ))[0]
print(f'Verified: feature_int_1={row[0]} (mutation), feature_int_2={row[1]} (DML) — both writes worked.')
```

| | Mutations API | DML |
|---|---|---|
| **What you send** | Buffered row changes (`insert`, `update`, `insert_or_update`, `delete`) | SQL (`INSERT` / `UPDATE` / `DELETE`, `INSERT OR UPDATE ... SELECT`) |
| **Atomicity** | Atomic within the commit (use `database.batch()` or a read-write txn) | Atomic within the enclosing read-write transaction |
| **Mutation-limit interaction** | Each column value + each index entry = 1 mutation; cap 80,000 per commit | Same 80,000 cap applies to rows the statement modifies |
| **Read-modify-write** | No — you must supply final values (blind write) | Yes — `SET x = x + 1`, or write rows computed by a `SELECT` |
| **When it wins** | Bulk loads, blind upserts, lowest write latency | Set-based updates, computed writes, conditional logic in one round trip |

**Interpreting:** for the feature-store ingest path, prefer the Mutations API — it skips SQL parse/plan and gives the highest write throughput. Reach for DML when the new value depends on the existing one or on a query (e.g. recomputing a feature from other columns server-side).

## 4. Range Scans and Batch Reads

Unlike Bigtable (where range scans require row key prefix design), Spanner supports range queries on any primary key column using standard SQL operators.

```python
# Range scan: entities 001000 through 001099 in group A
start = time.perf_counter()
with database.snapshot() as snapshot:
    results = list(snapshot.execute_sql(f"""
        SELECT entity_group, entity_id, feature_float_1, feature_string_1
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A'
          AND entity_id BETWEEN '001000' AND '001099'
        ORDER BY entity_id
    """))
range_ms = (time.perf_counter() - start) * 1000

print(f'Range scan (100 entities): {range_ms:.1f} ms, {len(results)} rows')
print(f'  First: {results[0][0]}/{results[0][1]}')
print(f'  Last:  {results[-1][0]}/{results[-1][1]}')
```

```python
# Batch read: specific entities by primary key
entity_ids = [f'{i:06d}' for i in [1, 100, 500, 1000, 2500, 5000]]
placeholders = ', '.join(f"('A', '{eid}')" for eid in entity_ids)

start = time.perf_counter()
with database.snapshot() as snapshot:
    results = list(snapshot.execute_sql(f"""
        SELECT entity_group, entity_id, feature_float_1, feature_string_1, label
        FROM {SPANNER_TABLE}
        WHERE (entity_group, entity_id) IN ({placeholders})
    """))
batch_ms = (time.perf_counter() - start) * 1000

print(f'Batch read ({len(entity_ids)} entities): {batch_ms:.1f} ms')
for row in results:
    print(f'  {row[0]}/{row[1]}: float_1={row[2]:.2f}, string_1={row[3]}, label={row[4]}')
```

## 5. Query Plans — EXPLAIN and PROFILE

Spanner exposes how it executes a query. A `PLAN`-mode request returns the operator tree without running the query; a `PROFILE`-mode request runs it and adds execution stats — most usefully **`rows_scanned`**. Because a Spanner table is physically stored sorted by its primary key, *the table is the primary index* — so a PK lookup and a non-key filter both appear as a `TableScan` in the plan tree. The difference is visible only in `rows_scanned`: a PK lookup seeks ~1 row, a non-key filter scans the whole table. That gap is what secondary indexes (Notebook 3) eliminate.


```python
from google.cloud.spanner_v1 import ExecuteSqlRequest

# Spanner stores a table physically sorted by its PRIMARY KEY, so there is no separate
# "primary index" object — a PK lookup is a TableScan with a single-row KEY SEEK, while a
# non-key filter is a TableScan that reads every row. In PLAN mode both look like a
# TableScan; the real difference shows up in PROFILE mode, which executes the query and
# reports rows_scanned. That gap (1 row vs 130,000) is the whole motivation for the
# secondary indexes added in Notebook 3.

def profile(sql, params=None, param_types=None):
    """Execute in PROFILE mode; return (operators, rows_returned, rows_scanned)."""
    with database.snapshot() as snapshot:
        rs = snapshot.execute_sql(
            sql, params=params, param_types=param_types,
            query_mode=ExecuteSqlRequest.QueryMode.PROFILE,
        )
        rows = list(rs)            # PROFILE actually runs the query
        stats = rs.stats
    ops = []
    if stats and stats.query_plan:
        KEEP = ('Distributed Union', 'Cross Apply', 'Serialize Result',
                'Scan', 'Filter', 'Sort', 'Aggregate')
        for node in stats.query_plan.plan_nodes:
            name = node.display_name or ''
            if any(k in name for k in KEEP):
                meta = dict(node.metadata) if node.metadata else {}
                tgt = meta.get('scan_target', '')
                ops.append(f'{name}' + (f' [{tgt}]' if tgt else ''))
    qs = dict(stats.query_stats) if stats and stats.query_stats else {}
    rows_returned = qs.get('rows_returned', str(len(rows)))
    rows_scanned = qs.get('rows_scanned', 'n/a')
    return ops, rows_returned, rows_scanned

def show(title, sql, params=None, param_types=None):
    ops, ret, scanned = profile(sql, params, param_types)
    print(title)
    for o in ops:
        print(f'  - {o}')
    print(f'  rows_returned={ret}  rows_scanned={scanned}')
    print()

# Query 1: primary-key point read -> single-row KEY SEEK (scans ~1 row)
pk_sql = f"""SELECT entity_group, entity_id, feature_float_1
             FROM {SPANNER_TABLE}
             WHERE entity_group = @g AND entity_id = @e"""
show('PK point read (WHERE entity_group=@g AND entity_id=@e):', pk_sql,
     params={'g': 'A', 'e': '000001'},
     param_types={'g': spanner.param_types.STRING, 'e': spanner.param_types.STRING})

# Query 2: non-key filter -> full TableScan (scans every row)
scan_sql = f"""SELECT entity_group, entity_id, feature_float_1
               FROM {SPANNER_TABLE}
               WHERE feature_float_1 > 900"""
show('Non-key filter (WHERE feature_float_1 > 900):', scan_sql)

print('Both are TableScans on dense_features, but rows_scanned tells the story:')
print('the PK read seeks ~1 row; the non-key filter scans all 130,000. Notebook 3')
print('adds a secondary index so the second query seeks instead of scanning.')
```

**Reading a Spanner query plan.** The optimizer returns a tree of operators; these are the ones you will see most often in feature-store queries:

| Plan operator | What it means | Relative cost |
|---|---|---|
| **TableScan** with a key seek | Table is sorted by PK, so a PK predicate seeks directly to the matching key range | Low — `rows_scanned` ≈ rows returned |
| **TableScan** (full) | Reads every row to apply a non-key predicate | **High** — `rows_scanned` ≈ whole table |
| **Index Scan** (secondary index) | Seeks by an indexed non-key column (added in Notebook 3) | Low — `rows_scanned` ≈ rows matched |
| **Distributed Union** | Fans the query across table splits and merges results | Low/medium — parallel, normal for multi-split reads |
| **Filter** | Applies a predicate to rows already produced | Low, but a Filter *on top of a full TableScan* is the red flag |
| **Serialize Result** | Formats the final output rows | Negligible |

**Interpreting:** in `PLAN` mode both queries below look like a `TableScan` — the honest signal is `rows_scanned` from `PROFILE` mode. The PK read seeks ~1 row; the `feature_float_1 > 900` filter scans all 130,000. A secondary index (Notebook 3) turns that second query into an Index Scan that seeks instead of scanning.

## 6. Strong Reads vs Stale Reads

Spanner offers a latency lever that Bigtable doesn't: **stale reads**. A strong read guarantees you see the latest committed data (linearizable consistency). A stale read allows Spanner to serve from a local replica that may be slightly behind, reducing latency.

For feature serving, stale reads are often acceptable — features that were current 10 or 15 seconds ago are fine for most ML models.

| Read Type | Consistency | Latency | When to Use |
|---|---|---|---|
| **Strong** | Linearizable | Higher | When you need the absolute latest value |
| **Exact staleness** | At most N seconds old | Lower | Feature serving where slight staleness is OK |
| **Max staleness** | Spanner picks freshest available within bound | Lowest | Maximum flexibility for latency optimization |

```python
query = f"""
    SELECT entity_group, entity_id, feature_float_1, feature_string_1, label
    FROM {SPANNER_TABLE}
    WHERE entity_group = @group AND entity_id = @eid
"""
params = {'group': 'A', 'eid': '000001'}
param_types = {'group': spanner.param_types.STRING, 'eid': spanner.param_types.STRING}

n_reads = 20

# Strong reads
strong_latencies = []
for i in range(n_reads):
    eid = f'{(i % 5000) + 1:06d}'
    p = {'group': 'A', 'eid': eid}
    start = time.perf_counter()
    with database.snapshot(multi_use=True) as snapshot:
        list(snapshot.execute_sql(query, params=p, param_types=param_types))
    strong_latencies.append((time.perf_counter() - start) * 1000)

# Stale reads (15 seconds staleness)
stale_latencies = []
for i in range(n_reads):
    eid = f'{(i % 5000) + 1:06d}'
    p = {'group': 'A', 'eid': eid}
    start = time.perf_counter()
    with database.snapshot(exact_staleness=datetime.timedelta(seconds=15)) as snapshot:
        list(snapshot.execute_sql(query, params=p, param_types=param_types))
    stale_latencies.append((time.perf_counter() - start) * 1000)

print(f'Point read latency ({n_reads} reads each):\n')
print(f'{"":<20s} {"Strong":>10s} {"Stale (15s)":>12s}')
print(f'{"Min":.<20s} {min(strong_latencies):>9.1f}ms {min(stale_latencies):>11.1f}ms')
print(f'{"Median":.<20s} {sorted(strong_latencies)[n_reads//2]:>9.1f}ms {sorted(stale_latencies)[n_reads//2]:>11.1f}ms')
print(f'{"p95":.<20s} {sorted(strong_latencies)[int(n_reads*0.95)]:>9.1f}ms {sorted(stale_latencies)[int(n_reads*0.95)]:>11.1f}ms')
print(f'{"Max":.<20s} {max(strong_latencies):>9.1f}ms {max(stale_latencies):>11.1f}ms')
print(f'{"Mean":.<20s} {sum(strong_latencies)/len(strong_latencies):>9.1f}ms {sum(stale_latencies)/len(stale_latencies):>11.1f}ms')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(range(n_reads), strong_latencies, 'o-', label='Strong read', color='#EA4335', alpha=0.8, markersize=4)
ax.plot(range(n_reads), stale_latencies, 's-', label='Stale read (15s)', color='#34A853', alpha=0.8, markersize=4)
ax.set_xlabel('Read #', fontweight='bold')
ax.set_ylabel('Latency (ms)', fontweight='bold')
ax.set_title('Strong vs Stale Read Latency', fontsize=13, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Bounded Staleness and a Reused Snapshot

`exact_staleness` (above) reads at a fixed point in the past. **`max_staleness`** is more flexible: it lets Spanner pick the *freshest* timestamp it can serve locally as long as the data is no older than the bound — often the lowest-latency option because it can avoid a cross-replica round trip.

The other lever is the **snapshot itself**. A read-only snapshot pins a single timestamp; reusing one snapshot for several reads gives every read a *consistent* view and amortizes session/transaction setup. This is the real serving pattern when you fetch many features for one prediction request.

```python
# max_staleness: Spanner serves the freshest value within a 15s bound.
max_stale_lat = []
for i in range(n_reads):
    eid = f'{(i % 5000) + 1:06d}'
    p = {'group': 'A', 'eid': eid}
    start = time.perf_counter()
    with database.snapshot(max_staleness=datetime.timedelta(seconds=15)) as snapshot:
        list(snapshot.execute_sql(query, params=p, param_types=param_types))
    max_stale_lat.append((time.perf_counter() - start) * 1000)

ms_p50 = sorted(max_stale_lat)[n_reads // 2]
ms_p95 = sorted(max_stale_lat)[int(n_reads * 0.95)]
print(f'max_staleness (15s): p50 {ms_p50:.1f}ms, p95 {ms_p95:.1f}ms over {n_reads} reads')
```

```python
# Real serving pattern: one read-only snapshot, several reads at ONE timestamp.
# multi_use=True keeps the snapshot open across reads (all share one timestamp).
serving_eids = [f'{(i * 137 % 5000) + 1:06d}' for i in range(10)]

start = time.perf_counter()
with database.snapshot(multi_use=True,
                       exact_staleness=datetime.timedelta(seconds=15)) as snapshot:
    fetched = []
    for eid in serving_eids:
        row = list(snapshot.execute_sql(
            query,
            params={'group': 'A', 'eid': eid},
            param_types=param_types,
        ))
        fetched.append(row[0] if row else None)
reused_ms = (time.perf_counter() - start) * 1000

print(f'Reused snapshot: {len(serving_eids)} reads at one timestamp in {reused_ms:.1f}ms '
      f'({reused_ms / len(serving_eids):.1f}ms per read)')
print(f'  Fetched {sum(1 for f in fetched if f)} entities, all from the same consistent snapshot.')
```

**Interpreting:** `max_staleness` typically matches or beats `exact_staleness` at p50 because Spanner is free to serve from the closest replica. Reusing one `multi_use` snapshot for a batch of feature reads is the pattern to copy into a serving service: every feature in a prediction request sees the same point-in-time view, and you pay transaction setup once instead of per read. For most ML serving a 10–15s staleness bound is invisible to model quality while measurably trimming tail latency.

## 8. Latency Baseline — Scaling with Entity Count

Measure how latency scales as we read more entities in a single query. This establishes the baseline for comparison with Bigtable (Notebook 9 of that series) and BigQuery Storage API.

```python
entity_counts = [1, 5, 10, 50, 100, 500, 1000]
results_table = []

for n in entity_counts:
    # Build entity list
    entities = [f'{i:06d}' for i in range(1, n + 1)]

    if n <= 100:
        # Use IN clause for small batches
        placeholders = ', '.join(f"('A', '{eid}')" for eid in entities)
        query = f"""
            SELECT entity_group, entity_id, feature_float_1, feature_string_1, label
            FROM {SPANNER_TABLE}
            WHERE (entity_group, entity_id) IN ({placeholders})
        """
    else:
        # Use range for larger batches
        query = f"""
            SELECT entity_group, entity_id, feature_float_1, feature_string_1, label
            FROM {SPANNER_TABLE}
            WHERE entity_group = 'A' AND entity_id <= '{entities[-1]}'
        """

    # Run 3 times, take median
    latencies = []
    for _ in range(3):
        start = time.perf_counter()
        with database.snapshot() as snapshot:
            rows = list(snapshot.execute_sql(query))
        latencies.append((time.perf_counter() - start) * 1000)

    median_ms = sorted(latencies)[1]
    results_table.append((n, len(rows), median_ms))
    print(f'  {n:>5,} entities: {median_ms:>8.1f} ms ({len(rows)} rows)')

print(f'\nLatency scales sub-linearly with entity count — Spanner parallelizes across splits.')
```

### Throughput (QPS) at 100 PU

Latency tells you how fast *one* read is; **throughput** tells you how many concurrent reads the instance sustains. Here we open a thread pool and hammer point reads for a fixed wall-clock window at a few concurrency levels, then report the measured queries-per-second. All numbers are for this **100 PU regional** instance.

```python
import concurrent.futures

def point_read(eid):
    with database.snapshot() as snapshot:
        list(snapshot.execute_sql(
            query,
            params={'group': 'A', 'eid': eid},
            param_types=param_types,
        ))

def run_qps(n_workers, window_s=3.0):
    """Issue point reads from n_workers threads for window_s seconds; return measured QPS."""
    stop_at = time.perf_counter() + window_s
    counter = {'n': 0}
    lock = __import__('threading').Lock()

    def worker(wid):
        local = 0
        i = wid
        while time.perf_counter() < stop_at:
            point_read(f'{(i % 5000) + 1:06d}')
            local += 1
            i += n_workers
        with lock:
            counter['n'] += local

    actual_start = time.perf_counter()
    with concurrent.futures.ThreadPoolExecutor(max_workers=n_workers) as ex:
        list(ex.map(worker, range(n_workers)))
    elapsed = time.perf_counter() - actual_start
    return counter['n'] / elapsed, counter['n'], elapsed

print(f'Concurrent point-read throughput (100 PU regional):\n')
print(f'{"Workers":>8s} {"Reads":>10s} {"Window":>9s} {"QPS":>10s}')
qps_results = []
for w in [1, 8, 32]:
    qps, reads, elapsed = run_qps(w, window_s=3.0)
    qps_results.append((w, qps))
    print(f'{w:>8d} {reads:>10,} {elapsed:>7.1f}s {qps:>10,.0f}')
```

**Interpreting:** with one worker you measure the serial ceiling (~1 / per-read latency). As workers increase, QPS climbs roughly linearly until the instance's compute (PU) or the client/network becomes the bottleneck — the point where QPS stops scaling with workers is your effective ceiling at this size. Spanner sizing is **capacity-based**: 1000 PU = 1 node, and throughput scales with PU, so if you need 2× the QPS measured here you provision ~2× the PU. Use a number like this — measured against *your* schema and read shape — to back into the PU you need for a target serving QPS, rather than relying on a generic per-node figure. (These results are also client-bound by the notebook VM; a distributed load generator will push higher.)

## 9. Comparison to Bigtable

The Bigtable Feature Store series (Notebook 1) establishes these baselines for the same 130K-entity dataset:

| Operation | Bigtable | Spanner (this notebook) | Notes |
|---|---|---|---|
| Single entity by key | ~3–5ms | ~5–10ms (strong) | Bigtable wins on raw point-lookup latency |
| 100 entities by key | ~50–100ms | ~10–30ms | Spanner parallelizes across splits |
| Range scan (100 contiguous) | ~10–20ms | ~5–15ms | Both efficient for range scans |
| Filter by non-key column | N/A (impossible) | ~50–200ms (no index) | Spanner's fundamental advantage |
| Aggregation (COUNT, AVG) | N/A (read all + client) | ~20–50ms | SQL pushdown |
| Schema discovery | Custom `#schema` row | `INFORMATION_SCHEMA` | Built-in for Spanner |
| Data types | Bytes (serialize/deserialize) | Native types | No encoding/decoding overhead |
| Stale reads | N/A (always strong, single cluster) | Yes (configurable staleness) | Latency lever |


> _Latency figures in this table are **approximate / illustrative** — see the Bigtable and Valkey notebooks for measured values on each store._

**When to choose Spanner over Bigtable for feature stores:**
- You need to query by non-key columns (secondary indexes — Notebook 3)
- You need JOINs at serving time (Notebook 7)
- You need multi-row ACID transactions (Notebook 5)
- You want native types without serialization overhead
- Stale reads are acceptable and you want the latency benefit

**When to choose Bigtable:**
- Sub-5ms point lookups are critical
- You need sub-ms atomic read-modify-write (counters)
- Your access pattern is purely key-based
- You need multi-cluster replication with app profile routing
