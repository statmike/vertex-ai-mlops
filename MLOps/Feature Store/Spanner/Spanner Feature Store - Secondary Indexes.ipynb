# Spanner Feature Store — Secondary Indexes and Query Patterns

Spanner's superpower for feature stores: query by **any indexed column** at serving time. In Bigtable, if you didn't design the row key for a specific query, you can't do it without a full table scan or denormalization. Spanner secondary indexes solve this generically.

**What you'll learn:**
- Create secondary indexes on feature columns and query by non-primary-key columns at sub-10ms latency
- `STORING` clause for covering queries — proven via the query **plan**, not just a latency delta
- Composite indexes for multi-column queries (and when the leading column rule defeats them)
- `NULL_FILTERED` indexes: a measured correctness demo showing which queries they can and cannot serve
- Interleaved (local) indexes vs global indexes
- Read the optimizer's **operator tree** with `PLAN` mode: Index Scan vs Table Scan, Distributed Union, Cross Apply
- Index anti-patterns — where an index does **not** help
- Write amplification measured as a before/after benchmark, tied to the 80,000-mutation commit limit

Every latency benchmark in this notebook reports **p50 / p95 / p99** from a shared `percentiles()` helper, not a single median.

**Key difference from Bigtable:** Bigtable can only query by row key prefix. If you need to find entities where `feature_float_1 > 80.0`, you'd have to read all rows and filter client-side. With a Spanner secondary index, this is a sub-10ms operation.

**Prerequisites:** Run [Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) and [Fundamentals](./Spanner%20Feature%20Store%20-%20Fundamentals.ipynb) first.

````python
import os  # noqa
# Silence the Spanner client's built-in Cloud Monitoring metrics exporter
# (noisy on incomplete resource labels; unrelated to the notebook).
os.environ.setdefault("SPANNER_DISABLE_BUILTIN_METRICS", "true")
from google.cloud import spanner
from google.cloud.spanner_v1 import ExecuteSqlRequest
import time
import matplotlib.pyplot as plt
import numpy as np

spanner_client = spanner.Client(project=PROJECT_ID)
instance = spanner_client.instance(SPANNER_INSTANCE)
database = instance.database(SPANNER_DATABASE)

# --- Shared benchmarking helper: report a distribution, never a single number ---
def percentiles(xs):
    """Return (p50, p95, p99) of a list of latency samples (ms), nearest-rank."""
    s = sorted(xs)
    n = len(s)
    def pct(p):
        # nearest-rank: index of the ceil(p*n)-th sample
        k = max(0, min(n - 1, int(round(p * (n - 1)))))
        return s[k]
    return pct(0.50), pct(0.95), pct(0.99)

def time_query(sql, n=30, warmup=2):
    """Run sql n times against fresh snapshots; return list of latencies (ms)."""
    for _ in range(warmup):
        with database.snapshot(multi_use=True) as snapshot:
            list(snapshot.execute_sql(sql))
    lats = []
    for _ in range(n):
        start = time.perf_counter()
        with database.snapshot(multi_use=True) as snapshot:
            list(snapshot.execute_sql(sql))
        lats.append((time.perf_counter() - start) * 1000)
    return lats

# --- Query-plan helper: print the operator TREE from PLAN mode ---
def explain_plan(sql, label=None):
    """Execute sql in PLAN mode (no rows materialized) and print the operator tree.

    Walks result.stats.query_plan.plan_nodes. Spanner returns a flat node list;
    each node's child_links reference children by index, so we render it as a tree.
    """
    with database.snapshot(multi_use=True) as snapshot:
        rs = snapshot.execute_sql(sql, query_mode=ExecuteSqlRequest.QueryMode.PLAN)
        list(rs)  # drain to populate stats
        plan = rs.stats.query_plan
    nodes = list(plan.plan_nodes)
    if label:
        print(label)

    def short_meta(node):
        # Surface scan target / index name and any STORING note from node metadata.
        # metadata is a Struct-like map that may be empty or None; coerce to a dict.
        try:
            md = dict(node.metadata) if node.metadata else {}
        except (TypeError, ValueError):
            md = {}
        # Prefer the keys the optimizer uses to name what it scanned; fall back to all.
        preferred = ('scan_type', 'scan_target', 'Full scan', 'iterator_type')
        items = [(k, md[k]) for k in preferred if k in md]
        if not items:
            items = list(md.items())
        bits = [f'{k}={v}' for k, v in items]
        return ('  [' + ', '.join(bits) + ']') if bits else ''

    def walk(idx, depth):
        node = nodes[idx]
        name = node.display_name or '(node)'
        kind_name = node.kind.name if hasattr(node.kind, 'name') else str(node.kind)
        print(f'{"  " * depth}{name} ({kind_name}){short_meta(node)}')
        for link in node.child_links:
            walk(link.child_index, depth + 1)

    walk(0, 0)
    print()
    return nodes

# Verify data exists
with database.snapshot(multi_use=True) as snapshot:
    count = list(snapshot.execute_sql(f'SELECT COUNT(*) FROM {SPANNER_TABLE}'))[0][0]
print(f'Spanner: {SPANNER_INSTANCE}/{SPANNER_DATABASE}/{SPANNER_TABLE} ({count:,} rows)')
````

---
## 1. Baseline — No Index (Full Table Scan)
First, measure the cost of querying by a non-primary-key column **without** an index. This is what you'd get in Bigtable for any query that doesn't match the row key pattern.

````python
# Query by non-key column without index. Each query is timed n times and we
# report the full p50/p95/p99 distribution via the shared percentiles() helper.
queries_no_index = [
    ('feature_string_1 = cat_5', f"SELECT entity_group, entity_id FROM {SPANNER_TABLE} WHERE feature_string_1 = 'cat_5' LIMIT 100"),
    ('feature_float_1 > 90.0', f"SELECT entity_group, entity_id FROM {SPANNER_TABLE} WHERE feature_float_1 > 90.0 LIMIT 100"),
    ('feature_bool_1 = TRUE', f"SELECT entity_group, entity_id FROM {SPANNER_TABLE} WHERE feature_bool_1 = TRUE LIMIT 100"),
    ('label = positive', f"SELECT entity_group, entity_id FROM {SPANNER_TABLE} WHERE label = 'positive' LIMIT 100"),
]

print('Queries WITHOUT secondary indexes (full table scan):\n')
print(f'{"Query":.<42s} {"p50":>8s} {"p95":>8s} {"p99":>8s}')

baseline_latencies = {}
for desc, sql in queries_no_index:
    lats = time_query(sql, n=15, warmup=1)
    p50, p95, p99 = percentiles(lats)
    baseline_latencies[desc] = (p50, p95, p99)
    print(f'{desc:.<42s} {p50:>6.0f}ms {p95:>6.0f}ms {p99:>6.0f}ms')
````

## 2. Create Secondary Indexes
Create indexes on the columns we want to query by. Each index is built asynchronously — Spanner processes the existing data in the background while the table remains fully readable and writable.

````python
index_ddl = [
    f'CREATE INDEX idx_string_1 ON {SPANNER_TABLE} (feature_string_1)',
    f'CREATE INDEX idx_float_1 ON {SPANNER_TABLE} (feature_float_1)',
    f'CREATE INDEX idx_bool_1 ON {SPANNER_TABLE} (feature_bool_1)',
    f'CREATE INDEX idx_label ON {SPANNER_TABLE} (label)',
]

for ddl in index_ddl:
    index_name = ddl.split('INDEX ')[1].split(' ON')[0]
    try:
        operation = database.update_ddl([ddl])
        print(f'Creating {index_name}...', end=' ')
        operation.result()
        print('done.')
    except Exception as e:
        if 'Duplicate' in str(e) or 'already exists' in str(e):
            print(f'{index_name} already exists.')
        else:
            print(f'{index_name}: {e}')
````

````python
# Verify indexes exist
with database.snapshot() as snapshot:
    indexes = list(snapshot.execute_sql("""
        SELECT INDEX_NAME, INDEX_TYPE, IS_UNIQUE, INDEX_STATE
        FROM INFORMATION_SCHEMA.INDEXES
        WHERE TABLE_NAME = @table AND INDEX_TYPE != 'PRIMARY_KEY'
        ORDER BY INDEX_NAME
    """, params={'table': SPANNER_TABLE}, param_types={'table': spanner.param_types.STRING}))

print(f'Secondary indexes on {SPANNER_TABLE}:\n')
print(f'{"Index Name":.<30s} {"Type":>10s} {"Unique":>8s} {"State":>12s}')
for name, itype, unique, state in indexes:
    print(f'{name:.<30s} {itype:>10s} {str(unique):>8s} {state:>12s}')
````

---
## 3. Indexed Queries — The Speedup
Run the same queries from Section 1, now backed by secondary indexes.

````python
print('Queries WITH secondary indexes (speedup compares p50):\n')
print(f'{"Query":.<42s} {"noidx p50":>10s} {"idx p50":>9s} {"idx p95":>9s} {"idx p99":>9s} {"speedup":>8s}')

indexed_latencies = {}
for desc, sql in queries_no_index:
    lats = time_query(sql, n=30, warmup=2)
    p50, p95, p99 = percentiles(lats)
    indexed_latencies[desc] = (p50, p95, p99)
    base_p50 = baseline_latencies[desc][0]
    speedup = base_p50 / p50 if p50 > 0 else float('inf')
    print(f'{desc:.<42s} {base_p50:>8.0f}ms {p50:>7.1f}ms {p95:>7.1f}ms {p99:>7.1f}ms {speedup:>7.1f}x')
````

````python
# Visualize the speedup (p50 bars, p95 shown as an error-bar cap)
labels = list(baseline_latencies.keys())
baseline_p50 = [baseline_latencies[l][0] for l in labels]
indexed_p50 = [indexed_latencies[l][0] for l in labels]
indexed_p95 = [indexed_latencies[l][1] for l in labels]

fig, ax = plt.subplots(figsize=(12, 5))
x = np.arange(len(labels))
width = 0.35

bars1 = ax.bar(x - width/2, baseline_p50, width, label='No Index (full scan), p50', color='#EA4335', alpha=0.8)
bars2 = ax.bar(x + width/2, indexed_p50, width, label='With Index, p50', color='#34A853', alpha=0.8)
# overlay p95 caps on the indexed bars
ax.errorbar(x + width/2, indexed_p50,
            yerr=[np.zeros(len(labels)), np.array(indexed_p95) - np.array(indexed_p50)],
            fmt='none', ecolor='#137333', capsize=4, label='Indexed p95')

ax.set_ylabel('Latency (ms)', fontweight='bold')
ax.set_title('Secondary Index Impact on Query Latency (p50, p95 cap)', fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=15, ha='right')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')

for bar, val in zip(bars1, baseline_p50):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5, f'{val:.0f}', ha='center', fontsize=9)
for bar, val in zip(bars2, indexed_p50):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5, f'{val:.1f}', ha='center', fontsize=9)

plt.tight_layout()
plt.show()
````

## 4. STORING Clause — Covering Queries
A basic secondary index contains only the indexed column(s) plus the primary key. To retrieve other columns, Spanner must do a **base table lookup** (join the index back to the table). The `STORING` clause adds extra columns directly to the index, enabling **index-only queries** with no base table lookup.

```sql
-- Without STORING: index scan + base table lookup
CREATE INDEX idx_string_1 ON features (feature_string_1)

-- With STORING: index scan only (covering query)
CREATE INDEX idx_string_1_covering ON features (feature_string_1)
    STORING (feature_float_1, feature_int_1, label)
```

Trade-off: `STORING` increases index storage and write amplification, but eliminates the base table lookup for queries that only need the stored columns.

````python
# Create a covering index
covering_ddl = f"""
    CREATE INDEX idx_string_1_covering ON {SPANNER_TABLE} (feature_string_1)
    STORING (feature_float_1, feature_int_1, label)
"""

try:
    operation = database.update_ddl([covering_ddl])
    print('Creating idx_string_1_covering (with STORING)...', end=' ')
    operation.result()
    print('done.')
except Exception as e:
    if 'Duplicate' in str(e) or 'already exists' in str(e):
        print('idx_string_1_covering already exists.')
    else:
        print(f'Error: {e}')
````

````python
# Compare: basic index vs covering index.
#
# FORCE_INDEX syntax note: the table hint is written as `table@{FORCE_INDEX=idx}`
# in GoogleSQL. Inside a Python f-string the literal braces must be DOUBLED so the
# f-string emits single braces. The expression below renders to exactly:
#     dense_features@{FORCE_INDEX=idx_string_1_covering}
# which is the correct, optimizer-binding hint. (Verify with: print(query_covering))
query_covering = f"""
    SELECT entity_group, entity_id, feature_float_1, feature_int_1, label
    FROM {SPANNER_TABLE}@{{FORCE_INDEX=idx_string_1_covering}}
    WHERE feature_string_1 = 'cat_5'
    LIMIT 100
"""

query_basic = f"""
    SELECT entity_group, entity_id, feature_float_1, feature_int_1, label
    FROM {SPANNER_TABLE}@{{FORCE_INDEX=idx_string_1}}
    WHERE feature_string_1 = 'cat_5'
    LIMIT 100
"""

# Confirm the doubled-brace escaping rendered to single braces (the real hint syntax).
assert f'{SPANNER_TABLE}@{{FORCE_INDEX=idx_string_1_covering}}' in query_covering

basic_p = percentiles(time_query(query_basic, n=30, warmup=2))
cover_p = percentiles(time_query(query_covering, n=30, warmup=2))

print(f'{"":24s} {"p50":>8s} {"p95":>8s} {"p99":>8s}')
print(f'{"Basic (+ base lookup)":24s} {basic_p[0]:>6.1f}ms {basic_p[1]:>6.1f}ms {basic_p[2]:>6.1f}ms')
print(f'{"Covering (index-only)":24s} {cover_p[0]:>6.1f}ms {cover_p[1]:>6.1f}ms {cover_p[2]:>6.1f}ms')
print('\nSTORING eliminates the base table lookup when the query only needs stored columns.')
print('The PLAN below proves WHY: the covering index has no back-join to the base table.')
````

### STORING proven by the query plan

A latency delta alone doesn't prove the base table lookup was eliminated — it could be caching noise. The query **plan** is the proof. Compare the operator trees:

- **Basic index** (`idx_string_1`): an `Index Scan` produces primary keys, then a `Cross Apply` back-joins to `dense_features` (a `Table Scan` on the inner side) to fetch `feature_float_1`, `feature_int_1`, `label`. Two scans.
- **Covering index** (`idx_string_1_covering` with `STORING (...)`): a single `Index Scan` returns everything. No back-join. The stored columns are physically present in the index, confirmed against `INFORMATION_SCHEMA.INDEX_COLUMNS` below.


````python
# PLAN contrast: basic index (Index Scan + back-join) vs covering index (Index Scan only)
print('=' * 72)
explain_plan(query_basic,    label='BASIC INDEX (idx_string_1) -- expect Index Scan + Cross Apply / Table Scan:')
print('=' * 72)
explain_plan(query_covering, label='COVERING INDEX (idx_string_1_covering, STORING) -- expect Index Scan only:')

# Make the STORING coverage explicit: which columns physically live in the index?
with database.snapshot() as snapshot:
    cover_cols = list(snapshot.execute_sql(
        """
        SELECT COLUMN_NAME, COLUMN_ORDERING
        FROM INFORMATION_SCHEMA.INDEX_COLUMNS
        WHERE TABLE_NAME = @t AND INDEX_NAME = 'idx_string_1_covering'
        ORDER BY ORDINAL_POSITION
        """,
        params={'t': SPANNER_TABLE},
        param_types={'t': spanner.param_types.STRING},
    ))

print('Columns physically present in idx_string_1_covering:')
for name, ordering in cover_cols:
    # Key columns have an ordering (ASC/DESC); STORING payload columns have NULL ordering.
    role = f'KEY ({ordering})' if ordering else 'STORING (payload)'
    print(f'  {name:<22s} {role}')
````

## 5. Composite Indexes
Index on multiple columns for multi-predicate queries.

````python
# Create composite index
composite_ddl = f"""
    CREATE INDEX idx_group_float1 ON {SPANNER_TABLE} (entity_group, feature_float_1)
"""

try:
    operation = database.update_ddl([composite_ddl])
    print('Creating idx_group_float1...', end=' ')
    operation.result()
    print('done.')
except Exception as e:
    if 'Duplicate' in str(e) or 'already exists' in str(e):
        print('idx_group_float1 already exists.')
    else:
        print(f'Error: {e}')
````

````python
# Query using the composite index (entity_group, feature_float_1).
# FORCE_INDEX hint again uses doubled braces -> renders to @{FORCE_INDEX=idx_group_float1}.
composite_query = f"""
    SELECT entity_group, entity_id, feature_float_1
    FROM {SPANNER_TABLE}@{{FORCE_INDEX=idx_group_float1}}
    WHERE entity_group = 'A' AND feature_float_1 > 90.0
"""

with database.snapshot() as snapshot:
    n_rows = len(list(snapshot.execute_sql(composite_query)))

p50, p95, p99 = percentiles(time_query(composite_query, n=30, warmup=2))
print(f'Composite index query (group=A, float_1>90): {n_rows} rows')
print(f'  p50={p50:.1f}ms  p95={p95:.1f}ms  p99={p99:.1f}ms')
print('\nThe leading column (entity_group) is an equality predicate, so the index can')
print("seek directly to group A and range-scan float_1 > 90 within it. This is the")
print('best case for a composite index. See the anti-patterns table for the worst case.')
````

### Index anti-patterns — when an index does *not* help

Indexes are not free and they are not magic. An index speeds up a query only when the optimizer can turn the predicate into an efficient **seek** (jump to a key range). The cases below force a scan anyway, so the index costs write amplification and storage with no read benefit.

| Anti-pattern | Example | Why the index can't seek | What to do instead |
|---|---|---|---|
| **Leading-column skip** on a composite index | `idx_group_float1 (entity_group, feature_float_1)` queried with only `WHERE feature_float_1 > 90` (no `entity_group`) | A composite index is sorted by its **first** column. Without an equality (or range) on `entity_group`, the rows matching `feature_float_1 > 90` are scattered across every group — no contiguous range to seek. | Add a single-column index on `feature_float_1`, or reorder the composite so the most-queried column leads. |
| **Low-cardinality column** | index on `feature_bool_1` (2 distinct values) | A seek on `= TRUE` still matches ~50% of the table. The optimizer often prefers a table scan over an index scan that touches half the rows plus a back-join. | Don't index it alone. Use it as a **trailing** column in a composite (`(entity_group, feature_bool_1)`) or fold it into a `STORING` payload. |
| **Leading-wildcard `LIKE`** | `WHERE feature_string_1 LIKE '%cat'` | B-tree indexes are ordered by prefix. A leading `%` means the matching values are spread across the whole key space — no prefix to seek. (`LIKE 'cat%'` *can* seek.) | Use `SEARCH`/tokenized search indexes, or restructure to a prefix match. |
| **Function / expression on the column** | `WHERE LOWER(feature_string_1) = 'cat_5'` | The index stores raw column values, not `LOWER(...)`. The optimizer must evaluate the function per row → full scan. | Create a generated column and index that, or store the value already normalized. |
| **Low selectivity overall** | predicate matching most rows (e.g. `feature_float_1 > 0`) | If a query returns most of the table, an index scan + back-join is *slower* than a straight table scan. | Let it table-scan; don't add the index. |

The next cell demonstrates the leading-column skip live: forcing the composite index for a query that only filters on the second column still degrades to a near-full index scan.


````python
# Demonstrate the leading-column skip: filter only on the SECOND column of the composite.
# Even when forced onto idx_group_float1, the engine must scan the whole index because
# there is no equality on the leading column entity_group to seek with.
skip_query = f"""
    SELECT entity_group, entity_id, feature_float_1
    FROM {SPANNER_TABLE}@{{FORCE_INDEX=idx_group_float1}}
    WHERE feature_float_1 > 90.0
    LIMIT 100
"""
print('=' * 72)
explain_plan(skip_query, label='Composite index, filtering ONLY the 2nd column (leading-column skip):')
print('Note: the Index Scan covers the full key range -- no seek on entity_group.')
print('A single-column idx_float_1 would be the right tool for this predicate.')
````

## 6. Interleaved (Local) Indexes vs Global Indexes
Every index so far is a **global index**: a standalone, table-wide structure that Spanner shards across splits independently of the base table. A global index can answer a query that touches *any* part of the table, but reading it may fan out across many splits (you saw `Distributed Union` in the plans).

Spanner also supports **interleaved indexes** (a.k.a. *local* indexes). An interleaved index is physically co-located with a parent table's rows, scoped to one **entity group** (one parent-key prefix):

```sql
-- Interleaved index: the index rows for each parent live in the SAME split
-- as the parent row. Requires the base table to be INTERLEAVED IN PARENT some_parent.
CREATE INDEX idx_local_orders_by_date
    ON orders (customer_id, order_date)          -- must be prefixed by the parent key
    INTERLEAVE IN customers
```

The base table in this notebook, `dense_features`, is a **top-level table with no interleaved parent**, so we cannot build a real interleaved index on it here — attempting `INTERLEAVE IN <parent>` would fail because there is no parent table to interleave into. We therefore present the DDL and the decision rule rather than create a broken index. To build one end-to-end, see the **[Interleaved Tables](./Spanner%20Feature%20Store%20-%20Interleaved%20Tables.ipynb)** notebook, which creates the parent/child hierarchy this requires.

### When a local (interleaved) index beats a global one

| Dimension | Local / interleaved index | Global index |
|---|---|---|
| **Query scope** | Best when the query is **scoped to a single entity group** (a known parent key), e.g. "the latest events for *this* customer". The read stays on one split. | Best when the query spans the whole table, e.g. "all rows where `feature_string_1 = 'cat_5'` across every entity". |
| **Fan-out** | None — single-split read, no `Distributed Union` over many servers. Lower, more predictable latency. | Fans out across splits; latency grows with how widely the matching rows are spread. |
| **Leading key requirement** | The index key **must be prefixed by the parent's primary key**. You can't query it without that prefix. | No prefix requirement; any indexed column can lead. |
| **Transaction locality** | Writes to a parent + its interleaved index entries hit one split → cheaper, hotspot-resistant single-split commits. | Index entries can land on different splits than the base row → wider commit, more coordination. |
| **Feature-store fit** | Per-entity feature timelines, child feature tables co-located with their parent entity. | Cross-entity serving lookups by feature value (the dominant pattern in this notebook). |

Rule of thumb: if your query *always* knows the parent key and only needs that entity's data, an interleaved index gives single-split, hotspot-resistant reads. If you query across entities by feature value, you want a global index.

## 7. NULL_FILTERED Indexes
Some feature columns have NULL values (our dataset has ~5% NULLs in several columns). A `NULL_FILTERED` index skips NULL rows, reducing storage and improving scan performance when you never query for NULLs.

````python
# Check NULL rates in our feature columns
with database.snapshot() as snapshot:
    null_stats = list(snapshot.execute_sql(f"""
        SELECT
            COUNT(*) AS total,
            COUNTIF(feature_float_3 IS NULL) AS float_3_nulls,
            COUNTIF(feature_string_2 IS NULL) AS string_2_nulls,
            COUNTIF(feature_date_2 IS NULL) AS date_2_nulls
        FROM {SPANNER_TABLE}
    """))

total = null_stats[0][0]
print(f'NULL rates in 130K rows:')
print(f'  feature_float_3:  {null_stats[0][1]:,} ({null_stats[0][1]*100/total:.1f}%)')
print(f'  feature_string_2: {null_stats[0][2]:,} ({null_stats[0][2]*100/total:.1f}%)')
print(f'  feature_date_2:   {null_stats[0][3]:,} ({null_stats[0][3]*100/total:.1f}%)')
````

````python
# Create NULL_FILTERED index
null_filtered_ddl = f"""
    CREATE NULL_FILTERED INDEX idx_string_2_nf ON {SPANNER_TABLE} (feature_string_2)
"""

try:
    operation = database.update_ddl([null_filtered_ddl])
    print('Creating idx_string_2_nf (NULL_FILTERED)...', end=' ')
    operation.result()
    print('done.')
except Exception as e:
    if 'Duplicate' in str(e) or 'already exists' in str(e):
        print('idx_string_2_nf already exists.')
    else:
        print(f'Error: {e}')

print(f'\nNULL_FILTERED index skips ~{null_stats[0][2]:,} NULL rows ({null_stats[0][2]*100/total:.1f}% storage savings).')
````

### NULL_FILTERED correctness: which queries it can (and can't) serve

A `NULL_FILTERED` index physically omits every row whose indexed column is NULL. That saves storage, but it changes **query coverage**, not just size:

- A query that filters `feature_string_2 = 'somevalue'` can use the index — those rows are non-NULL and therefore present in the index.
- A query that filters `feature_string_2 IS NULL` **cannot** use the index, because the matching rows were deliberately excluded. Forcing it would return wrong (incomplete) results, so the optimizer falls back to a base-table scan.

This is the trade-off: `NULL_FILTERED` is the right choice only when you never need to *find* the NULL rows through that index. The PLAN below proves both directions — index scan for the equality predicate, table scan for `IS NULL`.


````python
# Pick a real non-NULL value of feature_string_2 so the equality query returns rows.
with database.snapshot() as snapshot:
    sample_val = list(snapshot.execute_sql(f"""
        SELECT feature_string_2 FROM {SPANNER_TABLE}
        WHERE feature_string_2 IS NOT NULL LIMIT 1
    """))[0][0]
print(f"Using non-NULL sample value: feature_string_2 = '{sample_val}'\n")

# (a) Equality on a non-NULL value -> CAN use idx_string_2_nf. We let the optimizer
#     choose (no hint) to show it picks the NULL_FILTERED index on its own.
q_eq = f"""
    SELECT entity_group, entity_id
    FROM {SPANNER_TABLE}
    WHERE feature_string_2 = @v
    LIMIT 50
"""
# explain_plan uses execute_sql without params, so inline the literal for the PLAN demo:
q_eq_literal = f"SELECT entity_group, entity_id FROM {SPANNER_TABLE} WHERE feature_string_2 = '{sample_val}' LIMIT 50"

# (b) IS NULL -> the NULL_FILTERED index excludes these rows, so it CANNOT serve them.
q_null = f"SELECT entity_group, entity_id FROM {SPANNER_TABLE} WHERE feature_string_2 IS NULL LIMIT 50"

print('=' * 72)
explain_plan(q_eq_literal, label="(a) WHERE feature_string_2 = '...' -- expect Index Scan on idx_string_2_nf:")
print('=' * 72)
explain_plan(q_null,       label='(b) WHERE feature_string_2 IS NULL -- expect Table Scan (NF index cannot serve it):')
print("If you FORCE idx_string_2_nf onto the IS NULL query it returns 0 rows -- the NULL")
print("rows simply are not in the index. That is why the optimizer refuses to use it.")
````

## 8. Query Plan Analysis — Reading the Operator Tree
The previous sections used `PLAN` mode to prove specific points. This section makes plan-reading a first-class skill, because the plan is the only reliable way to know *which* index the optimizer chose and *why* a query is slow.

`PLAN` mode (`ExecuteSqlRequest.QueryMode.PLAN`) returns the optimizer's plan **without executing the query or materializing rows** — it's free to run and safe in production. The plan arrives as a flat list of nodes (`result.stats.query_plan.plan_nodes`); each node links to its children by index, so the `explain_plan()` helper renders it as an indented tree printing each node's `display_name` and `kind`.

We run four contrasting shapes so you can recognize the signatures:

1. **PK point-read** — equality on the full primary key. The cheapest possible plan.
2. **Indexed non-key filter, optimizer's choice** (no hint) — should pick an Index Scan.
3. **Same query, `FORCE_INDEX`** — confirms the hint binds and matches the optimizer's choice.
4. **Non-indexed column** — degrades to a Table Scan; this is what every Bigtable non-key query looks like.

### What each operator costs

| Operator | What it does | Cost signal |
|---|---|---|
| **Distributed Union** | Fans the query out to every split/server that may hold matching rows, then merges. Always the root of a distributed query. | One per query is normal. A Distributed Union *under* a scan means the scan spans many splits — wide fan-out, higher latency. |
| **Serialize Result** | Formats the final rows for the client. | Always near the root; negligible. |
| **Index Scan** | Seeks/scans a secondary index. Look at `scan_target` for the index name. | Cheap **if** it's a seek over a small key range. A full-range Index Scan is little better than a table scan. |
| **Table Scan** | Reads the base table directly. | Expensive on large tables. Acceptable for a PK seek; a red flag for a filtered non-key query. |
| **Cross Apply** | For each row from the left input, runs the right sub-plan — this is the **back-join** from a non-covering index to the base table. | Multiplies work by the left row count. Eliminate it with a `STORING` covering index. |
| **Filter** | Applies a residual predicate the index/scan couldn't satisfy. | A Filter on top of a Table Scan = no usable index for that predicate. |


````python
# Walk the operator TREE for four contrasting query shapes using PLAN mode.
# Look for: 'Table Scan' vs 'Index Scan' (with the index name in its metadata),
# 'Cross Apply' (back-join), 'Distributed Union' (fan-out), and 'Serialize Result'.

plan_queries = [
    ('1) PK point-read (full primary key)',
     f"SELECT * FROM {SPANNER_TABLE} WHERE entity_group = 'A' AND entity_id = '000001'"),
    ('2) Indexed non-key filter -- optimizer chooses (NO hint)',
     f"SELECT entity_group, entity_id FROM {SPANNER_TABLE} WHERE feature_string_1 = 'cat_5' LIMIT 50"),
    ('3) Same filter -- FORCE_INDEX=idx_string_1 (confirm the hint binds)',
     f"SELECT entity_group, entity_id FROM {SPANNER_TABLE}@{{FORCE_INDEX=idx_string_1}} WHERE feature_string_1 = 'cat_5' LIMIT 50"),
    ('4) Non-indexed column -- expect Table Scan + Filter',
     f"SELECT entity_group, entity_id FROM {SPANNER_TABLE} WHERE feature_int_3 = 42 LIMIT 50"),
]

for label, sql in plan_queries:
    print('=' * 72)
    explain_plan(sql, label=label)

print('Compare (2) and (3): the optimizer-chosen plan should match the forced one,')
print('confirming idx_string_1 is what it picks unaided. Compare (2) vs (4): Index')
print('Scan vs Table Scan is the whole story of why secondary indexes matter.')
````

## 9. Write Amplification — Measured Before / After
Every secondary index is a second copy of (a slice of) your data that Spanner must keep transactionally consistent. When you write a row, Spanner doesn't just write the base row — it also writes one **index entry per index** whose columns changed. Those index entries are extra **mutations** in the same commit.

This matters for two reasons:

1. **Latency** — more index entries per commit means more work per write.
2. **The 80,000-mutation-per-commit limit** — Spanner caps a single commit at 80,000 mutations. A mutation is counted *per column per row, plus per index entry*. A batch of `B` rows on a table with `K` secondary indexes can consume on the order of `B × (columns + K)` mutations. Add enough indexes (or stored columns) and a previously-fine batch size can blow the limit. ([Spanner limits](https://cloud.google.com/spanner/quotas#limits-for))

Rather than assert "indexes slow writes," we **measure** it:

1. Drop every secondary index → time `N` single-row `insert_or_update` writes (the zero-index baseline).
2. Recreate exactly the same indexes → time the **identical** writes again.
3. Compare writes/sec and p50/p95.

The same `percentiles()` helper reports the distribution. We record the index DDL *before* dropping so we recreate precisely what existed.


````python
# --- Step 1: capture the exact DDL of every secondary index, then drop them all ---
# We reconstruct CREATE INDEX statements from INFORMATION_SCHEMA so we recreate
# precisely what existed (name, NULL_FILTERED flag, key columns + ordering, STORING).

def secondary_index_defs():
    with database.snapshot(multi_use=True) as snapshot:
        idx_meta = list(snapshot.execute_sql("""
            SELECT INDEX_NAME, IS_UNIQUE, IS_NULL_FILTERED
            FROM INFORMATION_SCHEMA.INDEXES
            WHERE TABLE_NAME = @t AND INDEX_TYPE = 'INDEX'
            ORDER BY INDEX_NAME
        """, params={'t': SPANNER_TABLE}, param_types={'t': spanner.param_types.STRING}))
        cols = list(snapshot.execute_sql("""
            SELECT INDEX_NAME, COLUMN_NAME, COLUMN_ORDERING, ORDINAL_POSITION
            FROM INFORMATION_SCHEMA.INDEX_COLUMNS
            WHERE TABLE_NAME = @t
            ORDER BY INDEX_NAME, ORDINAL_POSITION
        """, params={'t': SPANNER_TABLE}, param_types={'t': spanner.param_types.STRING}))

    key_cols, stored_cols = {}, {}
    for name, col, ordering, pos in cols:
        if ordering:  # key column (has ASC/DESC ordering); ORDINAL_POSITION not NULL
            key_cols.setdefault(name, []).append(f'{col} {ordering}')
        else:         # STORING payload (NULL ordering)
            stored_cols.setdefault(name, []).append(col)

    defs = []
    for name, is_unique, is_nf in idx_meta:
        unique = 'UNIQUE ' if is_unique else ''
        nf = 'NULL_FILTERED ' if is_nf else ''
        keys = ', '.join(key_cols.get(name, []))
        ddl = f'CREATE {unique}{nf}INDEX {name} ON {SPANNER_TABLE} ({keys})'
        stored = stored_cols.get(name)
        if stored:
            ddl += f' STORING ({", ".join(stored)})'
        defs.append((name, ddl))
    return defs

saved_index_defs = secondary_index_defs()
print(f'Captured {len(saved_index_defs)} secondary index definition(s):')
for name, ddl in saved_index_defs:
    print(f'  {ddl}')

# Drop them all (one atomic schema change).
if saved_index_defs:
    database.update_ddl([f'DROP INDEX {name}' for name, _ in saved_index_defs]).result()
print(f'\nDropped {len(saved_index_defs)} secondary indexes. Table now has 0 secondary indexes.')
````

````python
# --- Step 2: benchmark identical writes with 0 indexes, then recreate and re-benchmark ---
N_WRITES = 200
write_columns = ['entity_group', 'entity_id', 'feature_float_1', 'feature_string_1', 'updated_at']

def time_writes(n):
    """Time n single-row insert_or_update commits; return latencies (ms)."""
    lats = []
    for i in range(n):
        eid = f'{(i % 5000) + 1:06d}'
        values = [['A', eid, 50.0 + i * 0.1, f'cat_{i % 20}', spanner.COMMIT_TIMESTAMP]]
        start = time.perf_counter()
        with database.batch() as batch:
            batch.insert_or_update(table=SPANNER_TABLE, columns=write_columns, values=values)
        lats.append((time.perf_counter() - start) * 1000)
    return lats

# Baseline: 0 secondary indexes (we dropped them in the previous cell).
lats_0 = time_writes(N_WRITES)
p50_0, p95_0, _ = percentiles(lats_0)

# Recreate EXACTLY the indexes we captured, then re-run the identical writes.
if saved_index_defs:
    print(f'Recreating {len(saved_index_defs)} indexes...', end=' ')
    database.update_ddl([ddl for _, ddl in saved_index_defs]).result()
    print('done.')

with database.snapshot() as snapshot:
    n_idx = list(snapshot.execute_sql("""
        SELECT COUNT(*) FROM INFORMATION_SCHEMA.INDEXES
        WHERE TABLE_NAME = @t AND INDEX_TYPE = 'INDEX'
    """, params={'t': SPANNER_TABLE}, param_types={'t': spanner.param_types.STRING}))[0][0]

lats_n = time_writes(N_WRITES)
p50_n, p95_n, _ = percentiles(lats_n)

def wps(p50_ms):
    return 1000.0 / p50_ms if p50_ms > 0 else float('inf')

print(f'\nSingle-row insert_or_update, {N_WRITES} writes each, identical payload:\n')
print(f'{"indexes":>12s} {"p50":>9s} {"p95":>9s} {"writes/sec (1/p50)":>20s}')
print(f'{"0":>12s} {p50_0:>7.1f}ms {p95_0:>7.1f}ms {wps(p50_0):>18.0f}')
print(f'{str(n_idx):>12s} {p50_n:>7.1f}ms {p95_n:>7.1f}ms {wps(p50_n):>18.0f}')

overhead = (p50_n - p50_0) / p50_0 * 100 if p50_0 > 0 else 0
print(f'\nPer-index note: with {n_idx} indexes each write also writes up to {n_idx} index')
print(f'entries; p50 write latency rose ~{overhead:.0f}% vs the zero-index baseline.')
print('Each indexed column adds index-entry mutations to the commit, so wide batches')
print('approach the 80,000-mutation-per-commit limit faster as you add indexes.')
print('For feature stores (read-heavy, batch/streaming writes) this trade is usually worth it.')
````

## Summary

Secondary indexes are Spanner's fundamental advantage over Bigtable for feature stores:

| Capability | Without Index | With Index |
|---|---|---|
| Query by feature value | Full table scan (slow) | Index scan (sub-10ms) |
| Multi-column predicate | Full scan + client filter | Composite index scan (leading column must be constrained) |
| Covering query | N/A | `STORING` clause → index-only, no back-join (proven in the plan) |
| NULL handling | Scan all rows | `NULL_FILTERED` (skip NULLs) — but cannot serve `IS NULL` |
| Per-entity locality | N/A | Interleaved (local) index → single-split reads |

**In Bigtable**, these queries are simply impossible without reading all rows client-side or maintaining a separate denormalized table for each access pattern.

### Skills this notebook built

- Read the optimizer's operator tree with `PLAN` mode and recognize Index Scan vs Table Scan, Cross Apply (back-join), and Distributed Union (fan-out).
- Prove `STORING` eliminates the base table lookup from the plan, not just a latency delta.
- Measure write amplification as a before/after benchmark and tie it to the 80,000-mutation commit limit.
- Avoid the index anti-patterns: leading-column skip, low-cardinality columns, leading-wildcard `LIKE`.
- All latency reported as p50 / p95 / p99 via the shared `percentiles()` helper.
