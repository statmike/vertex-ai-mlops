# Spanner Feature Store — Synchronization

Keeping Spanner fresh from BigQuery. This notebook covers the spectrum of sync patterns — from one-time backfill to near-real-time continuous export — plus change streams for reverse sync back to BigQuery.

**What you'll learn:**
- Batch sync: read from BigQuery, write to Spanner using `INSERT OR UPDATE` (native idempotent upsert)
- One-time reverse-ETL: `EXPORT DATA OPTIONS(format='CLOUD_SPANNER', ...)` — BigQuery writes directly to Spanner
- Continuous queries: the same `EXPORT DATA` statement run as a `CONTINUOUS` job for second-level freshness
- Reservation management: the Enterprise reservation + `CONTINUOUS` assignment that `EXPORT DATA` to Spanner requires
- Scheduled sync: a real BigQuery scheduled query (Data Transfer Service) that recomputes features on a cron
- Change streams: Spanner → Dataflow → BigQuery for reverse sync, including reading change records
- Error handling & idempotency: `INSERT OR UPDATE` under repeated apply, aborted-transaction retry, the 80,000-mutation commit limit
- The freshness trade-off: cost vs latency for each pattern

**Spanner's native reverse-ETL.** Just like Bigtable uses `EXPORT DATA OPTIONS(format='CLOUD_BIGTABLE', ...)`, Spanner is a first-class `EXPORT DATA` target with `format='CLOUD_SPANNER'`. A single BigQuery SQL statement writes query results straight into a Spanner table — no application read/write loop required. ([export-to-spanner](https://cloud.google.com/bigquery/docs/export-to-spanner))

**What makes the Spanner path distinct** (not better, just different from Bigtable):
- `spanner_options.change_timestamp_column` — names a `TIMESTAMP` column so continuous exports apply rows in commit order, which matters when the same key is updated repeatedly.
- The Spanner write path uses **`INSERT OR UPDATE`** upsert semantics natively — re-exporting an existing key updates it in place; there is no `overwrite` flag to manage.
- `spanner_options.priority` (`LOW`/`MEDIUM`/`HIGH`) lets sync traffic yield to latency-sensitive serving reads.
- Reverse sync uses **change streams → Dataflow → BigQuery** (Bigtable uses change streams → Pub/Sub).

**Prerequisites:** Run [Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) and [Fundamentals](./Spanner%20Feature%20Store%20-%20Fundamentals.ipynb) first.

**Sources:** [Export to Spanner (reverse ETL)](https://cloud.google.com/bigquery/docs/export-to-spanner) · [Continuous queries](https://cloud.google.com/bigquery/docs/continuous-queries) · [Spanner change streams](https://cloud.google.com/spanner/docs/change-streams) · [Change streams to BigQuery Dataflow template](https://cloud.google.com/dataflow/docs/guides/templates/provided/cloud-spanner-change-streams-to-bigquery) · [Spanner quotas & limits](https://cloud.google.com/spanner/quotas)

## 1. Sync Pattern Comparison

Patterns for keeping Spanner in sync with BigQuery, each with different freshness and cost characteristics. Spanner is a native `EXPORT DATA` target, so the same SQL statement serves both the one-time and continuous variants.

| Pattern | Mechanism | Freshness | Cost | Use Case |
|---|---|---|---|---|
| **Batch (`INSERT OR UPDATE`)** | Client-side read from BQ + Spanner mutations | Minutes–hours | Lowest | Small/targeted upserts, app-controlled transforms |
| **One-time `EXPORT DATA`** | `format='CLOUD_SPANNER'` reverse-ETL, one SQL statement | Minutes | Low (per-job slots) | Initial load, full/partial backfill |
| **Scheduled query** | Data Transfer Service runs SQL on a cron | Minutes | Low | Periodic feature recompute (hourly, daily) |
| **Continuous query** | `EXPORT DATA` run as a `CONTINUOUS` job over `APPENDS()` | Seconds | Higher (always-on slots) | Near-real-time feature propagation |

```
Freshness:
Hours              Minutes           Seconds           Milliseconds
|------------------|-----------------|-----------------|
Batch / EXPORT      Scheduled          Continuous         Direct writes
DATA backfill       queries            EXPORT DATA        (Notebook 5)
```

`EXPORT DATA` (one-time and continuous) and the scheduled query all run inside BigQuery — no application read/write loop. The client-side `INSERT OR UPDATE` path (Section 2) stays useful for small, surgical upserts and for transforms easier to express in Python than SQL.

## 2. Batch Sync — INSERT OR UPDATE

`INSERT OR UPDATE` is Spanner's native idempotent upsert. It inserts a row if the key doesn't exist, or updates the specified columns if it does. This makes client-side batch sync simple and safe to retry.

This client-side path (read from BigQuery in Python, write Spanner mutations) is **not** the only option — Section 3 shows the native one-statement `EXPORT DATA` reverse-ETL. Reach for client-side `INSERT OR UPDATE` when you want fine-grained control: small targeted upserts, custom transforms, or per-row error handling. The same `INSERT OR UPDATE` upsert semantics are what `EXPORT DATA` applies on the Spanner side, so the two paths converge on identical write behavior.

````python
# Demonstrate INSERT OR UPDATE with a test entity
test_entity = {
    'entity_group': 'A',
    'entity_id': '000001',
}

# Read current value
with database.snapshot(multi_use=True) as snapshot:
    before = list(snapshot.execute_sql(f"""
        SELECT feature_float_1, updated_at
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id = '000001'
    """))

print(f'Before: feature_float_1 = {before[0][0]:.4f}, updated_at = {before[0][1]}')

# Upsert with new value
new_float_value = 99.9999
with database.batch() as batch:
    batch.insert_or_update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
        values=[['A', '000001', new_float_value, spanner.COMMIT_TIMESTAMP]],
    )

# Read updated value
with database.snapshot(multi_use=True) as snapshot:
    after = list(snapshot.execute_sql(f"""
        SELECT feature_float_1, updated_at
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id = '000001'
    """))

print(f'After:  feature_float_1 = {after[0][0]:.4f}, updated_at = {after[0][1]}')
print(f'\nINSERT OR UPDATE is idempotent — safe to retry on failure.')
````

````python
# Restore original value
with database.batch() as batch:
    batch.insert_or_update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
        values=[['A', '000001', before[0][0], spanner.COMMIT_TIMESTAMP]],
    )
print('Original value restored.')
````

### Batch Sync Performance

Measure the throughput of batch INSERT OR UPDATE for different batch sizes.

````python
# Read a small set of entities from BigQuery
sample_df = bq.query(f'''
    SELECT entity_group, entity_id, feature_float_1, feature_float_2, feature_string_1, label
    FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
    WHERE entity_group = 'Z'
    LIMIT 1000
''').to_dataframe()

def to_spanner_values(row):
    return [
        row['entity_group'], row['entity_id'],
        float(row['feature_float_1']), float(row['feature_float_2']),
        row['feature_string_1'], row['label'],
        spanner.COMMIT_TIMESTAMP,
    ]

spanner_rows = [to_spanner_values(row) for _, row in sample_df.iterrows()]
columns = ['entity_group', 'entity_id', 'feature_float_1', 'feature_float_2',
           'feature_string_1', 'label', 'updated_at']

# Test different batch sizes
batch_sizes = [50, 100, 200, 500]
print(f'Batch sync performance (1,000 rows from group Z):\n')
print(f'{"Batch Size":>12s} {"Time (s)":>10s} {"Rows/sec":>10s}')

for bs in batch_sizes:
    start = time.perf_counter()
    for i in range(0, len(spanner_rows), bs):
        batch_rows = spanner_rows[i:i + bs]
        with database.batch() as batch:
            batch.insert_or_update(
                table=SPANNER_TABLE,
                columns=columns,
                values=batch_rows,
            )
    elapsed = time.perf_counter() - start
    rps = len(spanner_rows) / elapsed
    print(f'{bs:>12,} {elapsed:>10.2f} {rps:>10.0f}')
````

## 3. Reverse-ETL with EXPORT DATA — One-Time and Continuous

Spanner is a native [`EXPORT DATA`](https://cloud.google.com/bigquery/docs/export-to-spanner) target. A single BigQuery statement writes query results directly into a Spanner table using `format='CLOUD_SPANNER'` — exactly the pattern the Bigtable series uses with `format='CLOUD_BIGTABLE'`:

```sql
EXPORT DATA OPTIONS (
  uri='https://spanner.googleapis.com/projects/PROJECT/instances/INSTANCE/databases/DATABASE',
  format='CLOUD_SPANNER',
  spanner_options='''{"table": "dense_features"}'''
) AS
SELECT entity_group, entity_id, feature_float_1, ... FROM `project.dataset.dense_features`
```

The Spanner write path uses `INSERT OR UPDATE` semantics: re-exporting an existing primary key updates that row in place. There is no `overwrite` option (unlike Bigtable) — upsert is the only behavior.

### `spanner_options` fields

| Field | Purpose |
|---|---|
| `table` | Destination table (must already exist) |
| `change_timestamp_column` | A `TIMESTAMP` column used to apply rows in commit order — recommended for continuous exports where the same key is updated repeatedly |
| `priority` | Spanner write priority: `LOW`, `MEDIUM` (default), `HIGH` — set `LOW` so sync yields to serving reads |
| `tag` | Request tag for Spanner monitoring (default `bq_export`) |

### One-time vs continuous — same statement, different job type

The identical `EXPORT DATA` statement becomes a **continuous query** when submitted as a `CONTINUOUS` job over an `APPENDS()` source. It then runs indefinitely, exporting each new BigQuery row to Spanner within seconds.

```
BigQuery table (source)
    |
    +-- New row inserted (streaming insert / DML)
    |
    +-- APPENDS() surfaces the new row to the running CONTINUOUS job
           |
           +-- EXPORT DATA (format='CLOUD_SPANNER') upserts it into Spanner
                  |
                  +-- Row available for serving (~seconds)
```

### Continuous-query limitations (verified against the docs)

EXPORT-to-Spanner as a continuous query carries real constraints ([export-to-spanner](https://cloud.google.com/bigquery/docs/export-to-spanner), [continuous-queries](https://cloud.google.com/bigquery/docs/continuous-queries)):

- **Enterprise edition + reservation required.** Standard edition and on-demand compute cannot export to Spanner. The continuous variant additionally needs a `CONTINUOUS`-type reservation assignment (Section 3a).
- **No PostgreSQL-dialect target.** The destination must be a GoogleSQL-dialect Spanner database.
- **No auto-generated primary keys** on the destination table.
- **Hotspotting caution:** the docs warn against destination primary keys that aren't a monotonically increasing prefix — but the *opposite* caution applies for Spanner *write* hotspots. This series' composite key `(entity_group, entity_id)` is deliberately **non-monotonic** in its prefix (`entity_group` is a low-cardinality category, not a timestamp), so writes scatter across the key space and avoid a single hot split. That is the right shape for a Spanner sync target.
- **Source must use `APPENDS()`** with a start time inside the table's time-travel window (default 7 days).
- **Max row size 1 MiB**; a continuous job runs up to ~150 days under a service account.

````python
# The REAL continuous-query statement: BigQuery -> Spanner via EXPORT DATA.
# Submit with: bq query --use_legacy_sql=false --continuous=true '<this SQL>'
# (requires an Enterprise reservation + CONTINUOUS assignment; see Section 3a).
continuous_export_sql = f"""
EXPORT DATA OPTIONS (
  uri='{SPANNER_URI}',
  format='CLOUD_SPANNER',
  spanner_options='''{{"table": "{SPANNER_TABLE}", "change_timestamp_column": "updated_at", "priority": "LOW"}}'''
) AS
SELECT
  entity_group,
  entity_id,
  feature_float_1,
  feature_float_2,
  feature_string_1,
  label,
  CURRENT_TIMESTAMP() AS updated_at
FROM APPENDS(
  TABLE `{BQ_PROJECT}.{BQ_DATASET}.dense_features`,
  -- start at: now minus a few minutes, inside the time-travel window
  TIMESTAMP_SUB(CURRENT_TIMESTAMP(), INTERVAL 10 MINUTE)
)
"""

print('Continuous query (BigQuery -> Spanner reverse-ETL):')
print(continuous_export_sql)
print('change_timestamp_column=updated_at ensures rows apply in commit order;')
print('priority=LOW keeps sync writes behind latency-sensitive serving reads.')
````

### 3a. Reservation Management — Enterprise + CONTINUOUS Assignment

`EXPORT DATA` to Spanner runs only on **BigQuery Enterprise or Enterprise Plus** edition, which means a reservation with slots. The one-time export needs a `QUERY` assignment; the *continuous* variant needs a `CONTINUOUS` assignment.

**Cost note:** an Enterprise reservation bills its slots per second for as long as it exists. A `CONTINUOUS` assignment keeps slots always-on, so it is materially more expensive than the on-demand or scheduled paths — provision it only when you need second-level freshness, and delete it when you don't. The mirror of this section in the Bigtable Sync notebook creates the same Enterprise reservation for `EXPORT DATA`.

````python
from google.api_core.exceptions import AlreadyExists

# IMPORTANT: a QUERY and a CONTINUOUS assignment cannot live in the SAME reservation,
# so production uses one reservation per job type. EXPORT DATA to Spanner also requires
# Enterprise/Enterprise Plus edition. This cell only ACTS when RUN_RESERVATION=True;
# by default it just prints the commands so the notebook costs nothing to run.

parent = f'projects/{PROJECT_ID}/locations/{BQ_LOCATION}'

def reservation_cli():
    print('Equivalent bq CLI (one reservation per job type):')
    print(f'  bq mk --reservation --location={BQ_LOCATION} --slots={RESERVATION_SLOTS} '
          f'--edition=ENTERPRISE {RESERVATION_ID}')
    print(f'  bq mk --reservation_assignment --reservation_id={RESERVATION_ID} '
          f'--location={BQ_LOCATION} --job_type=QUERY --assignee_type=PROJECT --assignee_id={PROJECT_ID}')
    print('  # continuous variant uses a SEPARATE reservation with --job_type=CONTINUOUS')

if not RUN_RESERVATION:
    print('RUN_RESERVATION=False - not creating a billable reservation (default).')
    print('EXPORT DATA to Spanner needs a BigQuery Enterprise reservation; the one-time')
    print('and continuous-query cells below print the prerequisite if it is absent.')
    print()
    reservation_cli()
else:
    try:
        try:
            reservation = reservation_client.create_reservation(
                parent=parent, reservation_id=RESERVATION_ID,
                reservation=bigquery_reservation_v1.Reservation(
                    slot_capacity=RESERVATION_SLOTS,
                    edition=bigquery_reservation_v1.Edition.ENTERPRISE,
                ),
            )
            print(f'Created reservation: {reservation.name} ({RESERVATION_SLOTS} slots, ENTERPRISE)')
        except AlreadyExists:
            reservation = reservation_client.get_reservation(
                name=f'{parent}/reservations/{RESERVATION_ID}')
            print(f'Reservation exists: {reservation.name}')
        try:
            reservation_client.create_assignment(
                parent=reservation.name,
                assignment=bigquery_reservation_v1.Assignment(
                    job_type=bigquery_reservation_v1.Assignment.JobType.QUERY,
                    assignee=f'projects/{PROJECT_ID}'),
            )
            print('Created assignment: QUERY -> project')
        except AlreadyExists:
            print('Assignment exists: QUERY')
        print()
        print('Reservation ready (assignments take a few minutes to propagate).')
        print('Run the cleanup cell at the end (or cleanup.py) to stop billing.')
    except Exception as e:
        print(f'Could not create reservation: {type(e).__name__}: {str(e)[:200]}')
        reservation_cli()
````

### 3b. The Propagation Race — Executable End-to-End

This runs the **real** reverse-ETL, not a simulation:

1. Stream-insert a fresh row into a small BigQuery table via `insert_rows_json`.
2. Run a one-time `EXPORT DATA OPTIONS(format='CLOUD_SPANNER', ...)` to push it into Spanner.
3. Poll Spanner until the row appears and report end-to-end wall-clock latency.

`EXPORT DATA` to Spanner needs an Enterprise reservation (Section 3a). If one isn't available, the export step raises — we catch it and print the prerequisite, but the statement issued is the production-real one.

````python
# Executable propagation race: BQ streaming insert -> EXPORT DATA -> Spanner poll.
RACE_TABLE = f'{BQ_PROJECT}.{BQ_DATASET}.sync_race_temp'
RACE_GROUP, RACE_ID = 'A', '000001'
race_float = round(70.0 + np.random.random() * 10, 4)  # unique marker value

# 1) Ensure a tiny BQ source table exists, then stream-insert one row.
bq.query(f"""
    CREATE TABLE IF NOT EXISTS `{RACE_TABLE}` (
        entity_group STRING, entity_id STRING,
        feature_float_1 FLOAT64, feature_float_2 FLOAT64,
        feature_string_1 STRING, label STRING
    )
""").result()

t0 = time.perf_counter()
errors = bq.insert_rows_json(RACE_TABLE, [{
    'entity_group': RACE_GROUP, 'entity_id': RACE_ID,
    'feature_float_1': race_float, 'feature_float_2': 888.8888,
    'feature_string_1': 'cat_race_test', 'label': 'positive',
}])
assert not errors, errors
print(f'BQ streaming insert: marker feature_float_1={race_float}')

# 2) Run the REAL one-time EXPORT DATA to Spanner (Enterprise-gated).
export_sql = f"""
EXPORT DATA OPTIONS (
  uri='{SPANNER_URI}',
  format='CLOUD_SPANNER',
  spanner_options='{{"table": "{SPANNER_TABLE}", "priority": "LOW"}}'
) AS
SELECT entity_group, entity_id, feature_float_1, feature_float_2,
       feature_string_1, label, CURRENT_TIMESTAMP() AS updated_at
FROM `{RACE_TABLE}`
WHERE entity_group = '{RACE_GROUP}' AND entity_id = '{RACE_ID}'
  AND feature_float_1 = {race_float}
"""
try:
    bq.query(export_sql).result()
    # 3) Poll Spanner until the marker value lands.
    found, deadline = False, time.perf_counter() + 60
    while time.perf_counter() < deadline:
        with database.snapshot() as snap:
            rows = list(snap.execute_sql(
                f"SELECT feature_float_1 FROM {SPANNER_TABLE} "
                f"WHERE entity_group=@g AND entity_id=@e",
                params={'g': RACE_GROUP, 'e': RACE_ID},
                param_types={'g': spanner.param_types.STRING,
                             'e': spanner.param_types.STRING}))
        if rows and abs(rows[0][0] - race_float) < 1e-6:
            found = True
            break
        time.sleep(1)
    elapsed_ms = (time.perf_counter() - t0) * 1000
    if found:
        print(f'Marker visible in Spanner. End-to-end: {elapsed_ms:,.0f} ms')
    else:
        print(f'Marker not visible after 60s polling ({elapsed_ms:,.0f} ms elapsed).')
except Exception as e:
    print('PREREQUISITE: one-time EXPORT DATA to Spanner needs a BigQuery Enterprise')
    print('reservation (Section 3a). The statement issued was the production-real one:')
    print(f'  {type(e).__name__}: {str(e)[:300]}')
````

````python
# Verify the synced row in Spanner (state after the propagation race above).
with database.snapshot() as snapshot:
    result = list(snapshot.execute_sql(f"""
        SELECT entity_group, entity_id, feature_float_1, feature_float_2,
               feature_string_1, label, updated_at
        FROM {SPANNER_TABLE}
        WHERE entity_group = 'A' AND entity_id = '000001'
    """))

row = result[0]
print(f'Current Spanner state for A/000001:')
print(f'  feature_float_1:  {row[2]:.4f}')
print(f'  feature_string_1: {row[4]}')
print(f'  updated_at:       {row[6]}')
````

## 4. Scheduled Sync — Real BigQuery Scheduled Query

For features recomputed on a cadence (hourly, daily), the [BigQuery Data Transfer Service](https://cloud.google.com/bigquery/docs/scheduling-queries) runs SQL on a cron. The pattern: a scheduled query recomputes features into a BigQuery table, and either a `CONTINUOUS` `EXPORT DATA` job (Section 3) picks up the appends, or the scheduled SQL itself ends in an `EXPORT DATA OPTIONS(format='CLOUD_SPANNER', ...)` push to Spanner.

Below we create a **real** `TransferConfig` via `google.cloud.bigquery_datatransfer`. It's wrapped in try/except because it needs the Data Transfer API enabled and scheduling permissions.

````python
# Create a REAL BigQuery scheduled query (Data Transfer Service).
from google.cloud import bigquery_datatransfer_v1

# SQL the schedule runs: recompute features, then EXPORT DATA to Spanner.
SCHEDULED_SQL = f"""
CREATE OR REPLACE TABLE `{BQ_PROJECT}.{BQ_DATASET}.hourly_features` AS
SELECT
  entity_group, entity_id,
  ROUND(AVG(feature_float_1), 4) AS feature_float_1,
  ROUND(AVG(feature_float_2), 4) AS feature_float_2,
  ANY_VALUE(feature_string_1) AS feature_string_1,
  ANY_VALUE(label) AS label
FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
GROUP BY entity_group, entity_id;

EXPORT DATA OPTIONS (
  uri='{SPANNER_URI}',
  format='CLOUD_SPANNER',
  spanner_options='{{"table": "{SPANNER_TABLE}", "priority": "LOW"}}'
) AS
SELECT entity_group, entity_id, feature_float_1, feature_float_2,
       feature_string_1, label, CURRENT_TIMESTAMP() AS updated_at
FROM `{BQ_PROJECT}.{BQ_DATASET}.hourly_features`;
"""

transfer_client = bigquery_datatransfer_v1.DataTransferServiceClient()
transfer_parent = f'projects/{PROJECT_ID}/locations/{BQ_LOCATION}'
config = bigquery_datatransfer_v1.TransferConfig(
    display_name='Spanner Feature Store Sync — daily recompute',
    data_source_id='scheduled_query',
    params={'query': SCHEDULED_SQL},
    schedule='every 24 hours',  # e.g. 'every 6 hours', 'every day 02:00'
)
try:
    created = transfer_client.create_transfer_config(
        parent=transfer_parent, transfer_config=config)
    print(f'Created scheduled query: {created.name}')
    print(f'  Schedule: {created.schedule}')
    print('Delete with: transfer_client.delete_transfer_config(name=created.name)')
except Exception as e:
    print('PREREQUISITE: needs the BigQuery Data Transfer API enabled + scheduling perms,')
    print('and EXPORT DATA to Spanner needs an Enterprise reservation (Section 3a).')
    print(f'  {type(e).__name__}: {str(e)[:300]}')
    print('\nScheduled SQL that would run:')
    print(SCHEDULED_SQL)
````

## 5. Change Streams — Reverse Sync

[Spanner change streams](https://cloud.google.com/spanner/docs/change-streams) capture row-level mutations in real time. This enables reverse sync: when features are updated directly in Spanner (via the direct write path), change streams propagate those updates back to BigQuery for analytics and training data freshness.

```
Application → Direct write → Spanner
                                  │
                                  └─ Change stream → Dataflow → BigQuery
```

### Creating a Change Stream

````python
# Create a change stream on the feature table
change_stream_ddl = f"""
CREATE CHANGE STREAM feature_changes
  FOR {SPANNER_TABLE}
  OPTIONS (
    retention_period = '1d',
    value_capture_type = 'NEW_VALUES'
  )
"""

try:
    operation = database.update_ddl([change_stream_ddl])
    operation.result()
    print(f'Created change stream: feature_changes')
    print(f'  Retention: 1 day')
    print(f'  Captures: NEW_VALUES (post-mutation values)')
except Exception as e:
    if 'Duplicate' in str(e) or 'already exists' in str(e):
        print(f'Change stream feature_changes already exists.')
    else:
        print(f'Note: {e}')
````

````python
# Verify change stream exists
with database.snapshot() as snapshot:
    streams = list(snapshot.execute_sql("""
        SELECT CHANGE_STREAM_NAME, `ALL`
        FROM INFORMATION_SCHEMA.CHANGE_STREAMS
    """))

print('Change streams:')
for name, all_tables in streams:
    scope = 'all tables' if all_tables else 'specific tables'
    print(f'  {name} ({scope})')
````

### Reading Change Stream Records

A change stream is only useful once something **consumes** it. There are two consumption paths:

1. **Managed Dataflow template (production path).** The [`Spanner_Change_Streams_to_BigQuery`](https://cloud.google.com/dataflow/docs/guides/templates/provided/cloud-spanner-change-streams-to-bigquery) flex template reads the stream, manages partition tokens and checkpoints, and writes data-change records to BigQuery. This is the accurate, recommended reverse-sync path: **Spanner change stream → Dataflow → BigQuery.**
2. **Direct API read** via the change-stream read TVF (`READ_<stream>`) over `database.snapshot().execute_sql`, supplying partition tokens. This is lower-level and you own partition/checkpoint bookkeeping — fine for inspection, not for production.

Below we (a) make a mutation, (b) attempt a direct TVF read to surface real records, (c) print the production Dataflow command, and (d) decode the structure of a data-change record (`mod_type`, `keys`, `new_values`).

````python
import json as _json

# (a) Make a mutation that the change stream captures.
with database.batch() as batch:
    batch.update(
        table=SPANNER_TABLE,
        columns=['entity_group', 'entity_id', 'feature_float_1', 'updated_at'],
        values=[['A', '000002', 42.4242, spanner.COMMIT_TIMESTAMP]],
    )
print('Mutation applied to A/000002 (feature_float_1 = 42.4242).\n')

# (b) Direct API read of the change-stream TVF. The read TVF requires partition
# tokens; the initial query uses NULL to get the root partitions, then you read
# each child partition. We attempt a short read and surface any records.
start = (datetime.datetime.now(datetime.timezone.utc)
         - datetime.timedelta(seconds=30))
read_tvf = f"""
    SELECT ChangeRecord
    FROM READ_feature_changes(
        start_timestamp => @start,
        end_timestamp => NULL,
        partition_token => NULL,
        heartbeat_milliseconds => 10000
    )
"""
records_seen = 0
try:
    with database.snapshot() as snap:
        for row in snap.execute_sql(
                read_tvf,
                params={'start': start},
                param_types={'start': spanner.param_types.TIMESTAMP}):
            records_seen += 1
            if records_seen <= 1:
                print('Live change record (root partition response):')
                print(row[0])
            if records_seen >= 3:
                break
    if records_seen == 0:
        print('No records returned in the short window (partition reads are streamed; '
              'production uses Dataflow to manage tokens/checkpoints).')
except Exception as e:
    print('Direct TVF read needs partition-token bookkeeping; this is exactly what the')
    print(f'Dataflow template handles for you. ({type(e).__name__}: {str(e)[:160]})')

# (c) Production path: managed Dataflow flex template (Spanner -> BigQuery).
dataflow_cmd = f"""
gcloud dataflow flex-template run spanner-cs-to-bq-$(date +%s) \\
  --template-file-gcs-location=gs://dataflow-templates-{REGION}/latest/flex/Spanner_Change_Streams_to_BigQuery \\
  --region={REGION} \\
  --parameters \\
spannerInstanceId={SPANNER_INSTANCE},\\
spannerDatabase={SPANNER_DATABASE},\\
spannerMetadataInstanceId={SPANNER_INSTANCE},\\
spannerMetadataDatabase={SPANNER_DATABASE},\\
spannerChangeStreamName=feature_changes,\\
bigQueryDataset={BQ_DATASET}
""".strip()
print('\nProduction reverse-sync (Spanner change stream -> Dataflow -> BigQuery):')
print(dataflow_cmd)

# (d) Structure of a Spanner data-change record (NEW_VALUES capture type).
synthetic_record = {
    'mod_type': 'UPDATE',
    'table_name': SPANNER_TABLE,
    'keys': {'entity_group': 'A', 'entity_id': '000002'},
    'new_values': {'feature_float_1': 42.4242},
    'commit_timestamp': '2026-06-10T00:00:00.000000Z',
    'record_sequence': '00000000',
    'value_capture_type': 'NEW_VALUES',
}
print('\nData-change record structure (decoded):')
print(_json.dumps(synthetic_record, indent=2))
print('\nLoop closed: BigQuery -> Spanner (EXPORT DATA) and Spanner -> BigQuery (change stream -> Dataflow).')
````

## 7. Error Handling & Idempotency

Production sync must be safe under retries and partial failures. Spanner gives you three things to lean on:

- **`INSERT OR UPDATE` is idempotent.** Applying the same mutation twice produces the identical row — no duplicate rows, no double-counting. This is what makes both the client-side path and `EXPORT DATA` safe to retry.
- **Aborted transactions auto-retry.** Spanner may abort a transaction under contention; the client libraries retry the `run_in_transaction` body automatically with backoff. Keep transaction bodies side-effect-free so replay is safe.
- **80,000 mutations per commit.** A single commit is capped at 80,000 mutations, where each written column (plus every primary-key column, and each affected secondary-index entry) counts as one mutation. This bounds batch size: with a 7-column row you can write roughly 80,000 / 7 ≈ 11,400 rows per commit before splitting. ([Spanner quotas](https://cloud.google.com/spanner/quotas))

````python
# Idempotency: apply the SAME INSERT OR UPDATE twice -> identical result.
idem_cols = ['entity_group', 'entity_id', 'feature_float_1', 'updated_at']
idem_val = 55.5555

def apply_once():
    with database.batch() as batch:
        batch.insert_or_update(table=SPANNER_TABLE, columns=idem_cols,
                               values=[['A', '000001', idem_val, spanner.COMMIT_TIMESTAMP]])

def read_state():
    with database.snapshot() as snap:
        r = list(snap.execute_sql(
            f"SELECT feature_float_1, "
            f"(SELECT COUNT(*) FROM {SPANNER_TABLE} WHERE entity_group='A' AND entity_id='000001') "
            f"FROM {SPANNER_TABLE} WHERE entity_group='A' AND entity_id='000001'"))
    return r[0][0], r[0][1]

apply_once(); v1, n1 = read_state()
apply_once(); v2, n2 = read_state()
print(f'After 1st apply: feature_float_1={v1:.4f}, row_count={n1}')
print(f'After 2nd apply: feature_float_1={v2:.4f}, row_count={n2}')
print(f'Idempotent: {v1 == v2 and n1 == n2 == 1}  (same value, still exactly one row)')

# Mutation budget for this 7-column row.
COLS_PER_ROW = 7  # 5 written cols + 2 PK cols counted toward the 80,000 limit
print(f'\n80,000-mutation commit limit -> ~{80000 // COLS_PER_ROW:,} rows per commit '
      f'at {COLS_PER_ROW} mutations/row.')
````

## 8. Freshness Trade-offs

| Pattern | Freshness | Cost | Infra | Best For |
|---|---|---|---|---|
| **Client `INSERT OR UPDATE`** | Minutes–hours | Lowest | None | Targeted upserts, custom transforms |
| **One-time `EXPORT DATA`** | Minutes | Low (per-job slots) | BQ Enterprise reservation | Initial load, backfills |
| **Scheduled query** | Minutes | Low | Data Transfer Service (+reservation for the EXPORT step) | Periodic feature refresh |
| **Continuous `EXPORT DATA`** | Seconds | Higher (always-on slots) | Enterprise reservation + `CONTINUOUS` assignment | Near-real-time features |
| **Direct write** (Notebook 5) | Milliseconds | Per-write | None | Real-time counters, events |

**Compared to Bigtable sync patterns** — both stores are native `EXPORT DATA` targets; the differences are in semantics, not capability:

| | Bigtable | Spanner |
|---|---|---|
| **Reverse-ETL** | `EXPORT DATA` `format='CLOUD_BIGTABLE'` | `EXPORT DATA` `format='CLOUD_SPANNER'` |
| **Write semantics** | `overwrite=true` full replace, or upsert | `INSERT OR UPDATE` upsert (no `overwrite` flag) |
| **Continuous** | same `EXPORT DATA` as `CONTINUOUS` job | same `EXPORT DATA` as `CONTINUOUS` job |
| **Ordering control** | app-profile priority | `change_timestamp_column` + `priority` |
| **Reverse sync** | change streams → Pub/Sub → BQ | change streams → Dataflow → BQ |

### Spanner sync features without a Bigtable / Valkey analog

| Feature | What it does | Why it's Spanner-only here |
|---|---|---|
| **`INSERT OR UPDATE`** | Native row-level upsert | Idempotent merge on the relational row; Bigtable replaces cells, Valkey overwrites keys |
| **`change_timestamp_column`** | Applies continuous-export rows in commit order | Relational rows can be updated in place out-of-order; this serializes them |
| **Mutations vs DML write paths** | Mutations (batch API) skip the query layer for high-throughput bulk writes; DML runs inside a read-write transaction for conditional/computed updates | Two distinct write APIs with different cost/latency; KV stores have one path |
| **Request priority on sync jobs** | `spanner_options.priority=LOW` (and app-level priority) makes sync yield to serving reads | Per-request QoS on a strongly-consistent store |
