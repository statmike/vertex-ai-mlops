# Spanner Feature Store — Production Deployment and Capstone

Terraform deployment + a multi-table recommendation engine that demonstrates Spanner's unique value for feature serving: **ACID transactions, SQL JOINs at serving time, secondary indexes for complex predicates, and — the headline — a SQL metadata filter combined with native vector KNN in a single statement.**

The one thing no other store in this series can do: run a relational `WHERE` predicate *and* a vector nearest-neighbor ranking together, transaction-consistent, in one query — no client round-trip, no second system.

```sql
SELECT item_id, category, price, rating, COSINE_DISTANCE(embedding, @user_emb) AS dist
FROM item_catalog
WHERE category = @pref AND price < @budget AND rating > 3.5
ORDER BY dist
LIMIT 50
```

**What you'll learn:**
- Terraform split into `variables.tf` / `main.tf` / `outputs.tf` with a full `init → plan → apply → modify → destroy` lifecycle (autoscaling, deletion protection, IAM split, backup schedule)
- Multi-table recommendation engine: `user_profiles`, `item_catalog`, `user_item_interactions`
- **Native vector search**: `ARRAY<FLOAT32>(vector_length=>32)` columns, exact KNN, a ScaNN `VECTOR INDEX` for an ANN variant, and SQL-filtered KNN in one statement
- Serving-time JOINs across user, item, and interaction tables
- Transactional feature updates during serving
- A `FastAPI` `/recommend` + `/health` endpoint with a module-scoped client and parameterized queries (the prepared-statement analog)
- A real concurrent load test (`ThreadPoolExecutor`) reporting QPS and p50/p95/p99
- **Production operations**: Cloud Monitoring metrics + alerts, least-privilege IAM, HA/scaling, and a production-readiness checklist

**Capstone demonstrates what Bigtable's capstone genuinely cannot:**
- **Filter + vector KNN in one SQL statement** (Bigtable does brute-force KNN with no SQL filtering; Valkey filters via TAG/NUMERIC in a separate store)
- Serving-time JOINs: no denormalization needed
- ACID transactions: atomic multi-table updates during serving
- Typed DDL and secondary indexes: query by any indexed column

**Prerequisites:** Run [Environment](./Spanner%20Feature%20Store%20-%20Environment.ipynb) first. The [Vector Embeddings](./Spanner%20Feature%20Store%20-%20Vector%20Embeddings.ipynb) notebook covers native KNN/ANN in depth — this capstone wires it into a serving pipeline.

## 1. Terraform — Infrastructure as Code

[Terraform](https://www.terraform.io/) provisions the Spanner feature store declaratively: the instance (with autoscaling and deletion protection), the database and its schema, least-privilege IAM, a backup schedule, and point-in-time recovery. A production setup splits configuration across three files:

| File | Purpose |
|------|---------|
| `variables.tf` | Input parameters (project, region, instance name, autoscaling bounds) |
| `main.tf` | Resource definitions: instance, database + DDL, IAM split, backup schedule |
| `outputs.tf` | Values displayed after `apply` (instance/database names, serving SA) |

We write all three to disk, then walk the full **`init → plan → apply → modify → destroy`** lifecycle. Terraform may not run inside this notebook environment (no binary / no permission to create a parallel instance), so the lifecycle cells are presented as runnable shell that *also* prints the commands — the same pattern as the Bigtable sibling. The `.tf` files themselves are written and are directly usable from a shell.

````python
import platform, zipfile, urllib.request, stat, pathlib

TERRAFORM_VERSION = '1.15.2'
TF_DIR = 'terraform-spanner'
pathlib.Path(TF_DIR).mkdir(exist_ok=True)

arch = 'amd64' if platform.machine() == 'x86_64' else 'arm64'
tf_url = f'https://releases.hashicorp.com/terraform/{TERRAFORM_VERSION}/terraform_{TERRAFORM_VERSION}_linux_{arch}.zip'
tf_bin = pathlib.Path('terraform')
TF = None
try:
    if not tf_bin.exists():
        print(f'Downloading Terraform {TERRAFORM_VERSION} for linux_{arch}...')
        zip_path, _ = urllib.request.urlretrieve(tf_url)
        with zipfile.ZipFile(zip_path, 'r') as z:
            z.extract('terraform', '.')
        tf_bin.chmod(tf_bin.stat().st_mode | stat.S_IEXEC)
        pathlib.Path(zip_path).unlink()
    TF = str(tf_bin.resolve())
    out = subprocess.run([TF, 'version'], capture_output=True, text=True)
    print(out.stdout.strip())
except Exception as e:
    print(f'Terraform binary not available here ({e}).')
    print('The .tf files below are still written and runnable from a shell with terraform installed.')
````

### `variables.tf`

Inputs include **autoscaling bounds** (min/max processing units). On Spanner you choose either fixed capacity *or* an autoscaling policy — production almost always autoscales so capacity tracks load while staying within a budget ceiling.

````python
variables_tf = f"""variable "project_id" {{
  description = "GCP project ID"
  type        = string
  default     = "{PROJECT_ID}"
}}

variable "region" {{
  description = "GCP region for the regional instance config"
  type        = string
  default     = "{REGION}"
}}

variable "instance_name" {{
  description = "Spanner instance ID"
  type        = string
  default     = "{SPANNER_INSTANCE}"
}}

variable "database_name" {{
  description = "Spanner database ID"
  type        = string
  default     = "{SPANNER_DATABASE}"
}}

variable "min_processing_units" {{
  description = "Autoscaling floor (100 PU = 0.1 node)"
  type        = number
  default     = 100
}}

variable "max_processing_units" {{
  description = "Autoscaling ceiling — caps spend"
  type        = number
  default     = 1000
}}

variable "serving_sa" {{
  description = "Service account that performs serving-time reads"
  type        = string
  default     = "serving-sa@{PROJECT_ID}.iam.gserviceaccount.com"
}}

variable "sync_sa" {{
  description = "Service account that runs the BQ -> Spanner sync (writes)"
  type        = string
  default     = "sync-sa@{PROJECT_ID}.iam.gserviceaccount.com"
}}

variable "admin_sa" {{
  description = "Platform/admin service account (schema + instance management)"
  type        = string
  default     = "admin-sa@{PROJECT_ID}.iam.gserviceaccount.com"
}}
"""

(pathlib.Path(TF_DIR) / 'variables.tf').write_text(variables_tf)
print(variables_tf)
````

### `main.tf`

The instance uses an **`autoscaling_config`** (min/max PU + CPU/storage targets) instead of fixed `processing_units`, and sets **`deletion_protection = true`** so it can't be destroyed by accident. The database carries the recommendation DDL, **PITR** via `version_retention_period`, and a **least-privilege IAM split**: serving SA gets `databaseReader`, the sync SA gets `databaseUser`, the admin SA gets `databaseAdmin`. A `google_spanner_backup_schedule` adds daily backups.

> 65% high-priority CPU is Spanner's recommended autoscaling target for a **regional** instance (45% for multi-region, which must tolerate a region failure). We set `high_priority_cpu_utilization_percent = 65`.

````python
main_tf = r"""terraform {
  required_providers {
    google = {
      source  = "hashicorp/google"
      version = "~> 6.0"
    }
  }
}

provider "google" {
  project = var.project_id
  region  = var.region
}

# --- Spanner instance: autoscaling + deletion protection ---
resource "google_spanner_instance" "feature_store" {
  name         = var.instance_name
  config       = "regional-${var.region}"
  display_name = "Spanner Feature Store"
  edition      = "ENTERPRISE"   # required for ScaNN VECTOR INDEX (ANN)

  autoscaling_config {
    autoscaling_limits {
      min_processing_units = var.min_processing_units
      max_processing_units = var.max_processing_units
    }
    autoscaling_targets {
      high_priority_cpu_utilization_percent = 65   # regional recommendation
      storage_utilization_percent           = 95
    }
  }

  # Instances don't support `deletion_protection` (that's a database-level arg).
  # The instance guard is `force_destroy`: false (the default) blocks `terraform
  # destroy` while the instance still has databases. Set true only to allow teardown.
  force_destroy = false

  labels = {
    environment = "production"
    team        = "ml-platform"
  }
}

# --- Database: schema DDL + point-in-time recovery ---
resource "google_spanner_database" "features" {
  instance                 = google_spanner_instance.feature_store.name
  name                     = var.database_name
  version_retention_period = "7d"   # PITR window
  deletion_protection      = true

  ddl = [
    "CREATE TABLE user_profiles (user_id STRING(20) NOT NULL, age INT64, gender STRING(10), preferred_category STRING(20), avg_spend FLOAT64, lifetime_purchases INT64, embedding ARRAY<FLOAT32>(vector_length=>32) NOT NULL, updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)) PRIMARY KEY (user_id)",
    "CREATE TABLE item_catalog (item_id STRING(20) NOT NULL, category STRING(20), price FLOAT64, rating FLOAT64, popularity_score FLOAT64, embedding ARRAY<FLOAT32>(vector_length=>32) NOT NULL, updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)) PRIMARY KEY (item_id)",
    "CREATE TABLE user_item_interactions (user_id STRING(20) NOT NULL, item_id STRING(20) NOT NULL, interaction_type STRING(20), interaction_count INT64, last_interaction TIMESTAMP, updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)) PRIMARY KEY (user_id, item_id)",
    "CREATE INDEX idx_item_category ON item_catalog (category) STORING (price, rating, embedding)",
    "CREATE INDEX idx_item_rating ON item_catalog (rating)",
    "CREATE INDEX idx_item_price ON item_catalog (price)",
    "CREATE INDEX idx_interaction_item ON user_item_interactions (item_id)",
    "CREATE VECTOR INDEX idx_item_embedding ON item_catalog(embedding) OPTIONS (distance_type='COSINE', tree_depth=2, num_leaves=1000)",
  ]
}

# --- Daily backup schedule ---
resource "google_spanner_backup_schedule" "daily" {
  instance           = google_spanner_instance.feature_store.name
  database           = google_spanner_database.features.name
  name               = "daily-backup"
  retention_duration = "604800s"   # 7 days

  spec {
    cron_spec {
      text = "0 2 * * *"   # 02:00 daily
    }
  }

  full_backup_spec {}
}

# --- Least-privilege IAM: serving / sync / admin ---
resource "google_spanner_database_iam_member" "serving_reader" {
  instance = google_spanner_instance.feature_store.name
  database = google_spanner_database.features.name
  role     = "roles/spanner.databaseReader"
  member   = "serviceAccount:${var.serving_sa}"
}

resource "google_spanner_database_iam_member" "sync_user" {
  instance = google_spanner_instance.feature_store.name
  database = google_spanner_database.features.name
  role     = "roles/spanner.databaseUser"
  member   = "serviceAccount:${var.sync_sa}"
}

resource "google_spanner_database_iam_member" "admin" {
  instance = google_spanner_instance.feature_store.name
  database = google_spanner_database.features.name
  role     = "roles/spanner.databaseAdmin"
  member   = "serviceAccount:${var.admin_sa}"
}
"""

(pathlib.Path(TF_DIR) / 'main.tf').write_text(main_tf)
print(main_tf[:1400] + '\n... (full file written to main.tf)')
````

````python
outputs_tf = """output "instance_name" {
  value       = google_spanner_instance.feature_store.name
  description = "Spanner instance ID"
}

output "database_name" {
  value       = google_spanner_database.features.name
  description = "Spanner database ID"
}

output "serving_service_account" {
  value       = var.serving_sa
  description = "SA granted roles/spanner.databaseReader"
}

output "autoscaling_bounds" {
  value       = "${var.min_processing_units}-${var.max_processing_units} PU"
  description = "Autoscaling floor and ceiling"
}
"""

(pathlib.Path(TF_DIR) / 'outputs.tf').write_text(outputs_tf)
print(outputs_tf)
````

### Lifecycle: `init → plan`

The Terraform lifecycle is the same whether you provision one dev instance or a multi-region production deployment:

1. **`terraform init`** — downloads the Google provider plugin
2. **`terraform plan`** — dry run; shows exactly what will be created/changed/destroyed
3. **`terraform apply`** — executes the plan against real infrastructure

The cell below runs `init` then `plan` if the Terraform binary is available; otherwise it prints the commands to run from a shell. Nothing is created until `apply`.

````python
def run_tf(*args, timeout=300):
    """Run a terraform subcommand if the binary is available; else print the command."""
    if not TF:
        print('$ terraform ' + ' '.join(args) + '   (binary not available in this environment)')
        return None
    r = subprocess.run([TF, *args], cwd=TF_DIR, capture_output=True, text=True, timeout=timeout)
    out = r.stdout[-3000:] if len(r.stdout) > 3000 else r.stdout
    print(out)
    if r.returncode != 0:
        print('STDERR:', r.stderr[-1500:])
    return r

run_tf('init', '-no-color')
print('\n' + '=' * 60 + '\n')
run_tf('plan', '-no-color')
````

### `apply` — create the infrastructure

`terraform apply` creates the instance (1–2 min), the database with its full schema (including the ScaNN `VECTOR INDEX`, built as a background operation), the backup schedule, and the IAM bindings. With `deletion_protection = true`, a later `destroy` requires flipping that flag first — a deliberate guardrail.

> The rest of this notebook builds the same tables directly via the Python client so the capstone runs end-to-end against the existing series instance without requiring a parallel Terraform-managed instance. The Terraform path is the production deployment story; the Python path is the live demo.

````python
# Gated so the notebook does not create a parallel instance by default.
RUN_TF_APPLY = False

if RUN_TF_APPLY and TF:
    print('Applying Terraform configuration (creates a Spanner instance — may take minutes)...\n')
    run_tf('apply', '-auto-approve', '-no-color', timeout=600)
else:
    print('Skipped terraform apply (RUN_TF_APPLY = False).')
    print('To provision from a shell:')
    print(f'  cd {TF_DIR}')
    print('  terraform init && terraform plan && terraform apply')
````

### `modify` — evolve the infrastructure and review the plan diff

The production workflow for changing infrastructure: edit the `.tf` files, run `terraform plan` to review the exact diff, then `apply`. Here we **add a serving-replica IAM binding** for a second read service account (e.g., a regional batch-scoring job). `terraform plan` shows a single additive change — `1 to add, 0 to change, 0 to destroy` — the safe diff you want for an additive IAM grant.

````python
main_tf_path = pathlib.Path(TF_DIR) / 'main.tf'
current = main_tf_path.read_text()

iam_addition = """

# --- Added in modify step: second serving SA (regional batch scoring) ---
resource "google_spanner_database_iam_member" "batch_reader" {
  instance = google_spanner_instance.feature_store.name
  database = google_spanner_database.features.name
  role     = "roles/spanner.databaseReader"
  member   = "serviceAccount:batch-scoring-sa@${var.project_id}.iam.gserviceaccount.com"
}
"""

if 'batch_reader' not in current:
    main_tf_path.write_text(current + iam_addition)
    print('Modified main.tf — added batch-scoring serving reader (additive IAM grant).')
else:
    print('batch_reader already present in main.tf.')

print('\nReview the diff:')
run_tf('plan', '-no-color')
# Then apply (additive, zero-downtime).
````

### `destroy` — tear down

`terraform destroy` removes everything Terraform created — no orphaned resources. Because `deletion_protection = true` is set on both the instance and database, destroy fails until you set those to `false` and `apply` once (the guardrail working). For a managed teardown of *this series'* resources, use the `cleanup.py` helper in this folder — it does not touch a Terraform-managed instance.

````python
RUN_TF_DESTROY = False

if RUN_TF_DESTROY and TF:
    # Requires force_destroy=true (instance) + deletion_protection=false (database) applied first.
    run_tf('destroy', '-auto-approve', '-no-color', timeout=600)
else:
    print('Skipped terraform destroy (RUN_TF_DESTROY = False).')
    print('To tear down a Terraform-managed instance:')
    print('  1. Set force_destroy = true (instance) and deletion_protection = false (database) in main.tf')
    print('  2. terraform apply   (applies the flag change)')
    print('  3. terraform destroy')
````

## 2. Capstone: Multi-Table Recommendation Engine

Build a recommendation system on three Spanner tables that does what Bigtable's capstone cannot: **a SQL metadata filter combined with native vector KNN in one statement**, plus serving-time JOINs and atomic cross-table updates.

```
user_profiles                 item_catalog                    user_item_interactions
├─ user_id (PK)               ├─ item_id (PK)                 ├─ user_id (PK)
├─ demographics               ├─ category, price, rating       ├─ item_id (PK)
├─ preferred_category         ├─ popularity_score              ├─ interaction_type
└─ embedding                  └─ embedding                     └─ interaction_count
   ARRAY<FLOAT32>(32)            ARRAY<FLOAT32>(32)
                                 + VECTOR INDEX (ScaNN, COSINE)
```

The `embedding` columns are typed `ARRAY<FLOAT32>(vector_length=>32) NOT NULL` so they are eligible for Spanner's native distance functions and a ScaNN `VECTOR INDEX` (see the [Vector Embeddings](./Spanner%20Feature%20Store%20-%20Vector%20Embeddings.ipynb) notebook). Stage-1 candidate retrieval uses them directly.

### Create Recommendation Tables

````python
rec_ddl = [
    """CREATE TABLE IF NOT EXISTS user_profiles (
        user_id STRING(20) NOT NULL,
        age INT64,
        gender STRING(10),
        preferred_category STRING(20),
        avg_spend FLOAT64,
        lifetime_purchases INT64,
        embedding ARRAY<FLOAT32>(vector_length=>32) NOT NULL,
        updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
    ) PRIMARY KEY (user_id)""",

    """CREATE TABLE IF NOT EXISTS item_catalog (
        item_id STRING(20) NOT NULL,
        category STRING(20),
        price FLOAT64,
        rating FLOAT64,
        popularity_score FLOAT64,
        embedding ARRAY<FLOAT32>(vector_length=>32) NOT NULL,
        updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
    ) PRIMARY KEY (item_id)""",

    """CREATE TABLE IF NOT EXISTS user_item_interactions (
        user_id STRING(20) NOT NULL,
        item_id STRING(20) NOT NULL,
        interaction_type STRING(20),
        interaction_count INT64,
        last_interaction TIMESTAMP,
        updated_at TIMESTAMP NOT NULL OPTIONS (allow_commit_timestamp=true)
    ) PRIMARY KEY (user_id, item_id)""",

    # STORING(price, rating, embedding) -> filtered exact-KNN runs as an index-only scan
    "CREATE INDEX IF NOT EXISTS idx_item_category ON item_catalog (category) STORING (price, rating, embedding)",
    "CREATE INDEX IF NOT EXISTS idx_item_rating ON item_catalog (rating)",
    "CREATE INDEX IF NOT EXISTS idx_item_price ON item_catalog (price)",
    "CREATE INDEX IF NOT EXISTS idx_interaction_item ON user_item_interactions (item_id)",
]

try:
    operation = database.update_ddl(rec_ddl)
    operation.result()
    print('Recommendation tables created (embeddings: ARRAY<FLOAT32>(vector_length=>32)).')
except Exception as e:
    print(f'DDL: {e}')
````

### Build the ScaNN `VECTOR INDEX` (ANN variant)

For Stage-1 retrieval we will use **exact KNN** with `COSINE_DISTANCE` (precise, and ideal once a SQL filter has narrowed the candidate set). For the larger unfiltered case we also build a **ScaNN `VECTOR INDEX`** so the same query can run as approximate nearest-neighbor via `APPROX_COSINE_DISTANCE`.

The index build is an asynchronous schema operation — `update_ddl(...).result()` blocks until it completes. ANN requires Spanner **Enterprise / Enterprise Plus** edition, so we detect the edition and set `ANN_AVAILABLE` accordingly (exact KNN works on every edition).

````python
# Resolve the ARRAY<FLOAT32> query-param type (FLOAT64 fallback for older clients).
pt = spanner.param_types
if hasattr(pt, 'FLOAT32'):
    Q_ARRAY_TYPE = pt.Array(pt.FLOAT32)
    print('Query vectors: ARRAY<FLOAT32> param type.')
else:
    Q_ARRAY_TYPE = pt.Array(pt.FLOAT64)
    print('FLOAT32 scalar param type not in this client — using ARRAY<FLOAT64> (Spanner coerces).')

# Build the ScaNN VECTOR INDEX on item_catalog(embedding). Async schema op; .result() waits.
ANN_AVAILABLE = True
try:
    print('Creating VECTOR INDEX idx_item_embedding (background op; .result() waits)...')
    start = time.perf_counter()
    database.update_ddl([
        """CREATE VECTOR INDEX idx_item_embedding
               ON item_catalog(embedding)
               OPTIONS (distance_type='COSINE', tree_depth=2, num_leaves=1000)"""
    ]).result()
    print(f'  Vector index ready ({time.perf_counter()-start:.0f}s).')
except Exception as e:
    msg = str(e)
    if 'Duplicate' in msg or 'already exists' in msg:
        print('  idx_item_embedding already exists.')
    elif 'Enterprise' in msg or 'edition' in msg.lower() or 'not supported' in msg.lower():
        ANN_AVAILABLE = False
        print(f'  ANN not available on this instance edition — exact KNN still works.\n  ({msg[:140]})')
    else:
        raise
````

## 3. SQL-Powered Recommendation Pipeline

The pipeline uses SQL at every step — and the headline is **Stage 1**, where a relational filter and a vector nearest-neighbor ranking run in the *same* statement:

1. **Candidate retrieval**: `WHERE category=… AND price<… AND rating>…` **+ `ORDER BY COSINE_DISTANCE(embedding, @user_emb)`** in one query
2. **Feature assembly**: JOIN across user, item, and interaction tables
3. **Scoring**: a lightweight linear scorer on the assembled features (no model dependency)
4. **Update**: atomic cross-table interaction + popularity update

Bigtable's capstone can do brute-force KNN but cannot apply the SQL filter in the same step; Valkey filters via TAG/NUMERIC but in a separate store from the relational data. Spanner does both, transaction-consistent, in one round-trip.

````python
# Stage 1: SQL metadata filter + native vector KNN in ONE statement.
# "For this user's taste vector, find electronics under $500 rated > 3.5,
#  ranked by cosine distance to the user embedding."

# Pull the query user's embedding (the "taste vector").
user_id = 'user_000001'
with database.snapshot(multi_use=True) as snapshot:
    urow = list(snapshot.execute_sql(
        "SELECT preferred_category, avg_spend, embedding FROM user_profiles WHERE user_id = @uid",
        params={'uid': user_id}, param_types={'uid': spanner.param_types.STRING}))
pref_cat, avg_spend, user_emb = urow[0][0], urow[0][1], [float(x) for x in urow[0][2]]

start = time.perf_counter()
with database.snapshot(exact_staleness=datetime.timedelta(seconds=10)) as snapshot:
    candidates = list(snapshot.execute_sql("""
        SELECT item_id, category, price, rating,
               COSINE_DISTANCE(embedding, @user_emb) AS dist
        FROM item_catalog
        WHERE category = @pref AND price < @budget AND rating > 3.5
        ORDER BY dist
        LIMIT 50
    """, params={'user_emb': user_emb, 'pref': pref_cat, 'budget': avg_spend * 2.0},
         param_types={'user_emb': Q_ARRAY_TYPE,
                      'pref': spanner.param_types.STRING,
                      'budget': spanner.param_types.FLOAT64}))
candidate_ms = (time.perf_counter() - start) * 1000

print('Stage 1: SQL filter + native vector KNN (ONE statement)')
print(f'  User {user_id}: pref={pref_cat}, budget<${avg_spend*2:.0f}')
print(f'  Filter: category={pref_cat}, price<{avg_spend*2:.0f}, rating>3.5')
print(f'  Candidates returned: {len(candidates)}')
print(f'  Latency: {candidate_ms:.1f}ms')
print('\n  Nearest by cosine distance:')
for item_id, cat, price, rating, dist in candidates[:5]:
    print(f'    {item_id}: {cat}, ${price:.2f}, rating={rating}, cosine_dist={dist:.4f}')
print('\n  Bigtable: brute-force KNN only, no SQL filter in the same step.')
print('  Valkey:   TAG/NUMERIC filter + HNSW, but in a separate store.')
````

**ANN variant.** The query above is *exact* KNN — after the `WHERE` narrows to one category, scanning those rows is cheap and precise. For an *unfiltered* nearest-neighbor over the whole catalog, the ScaNN `VECTOR INDEX` lets the same ranking run as approximate NN via `APPROX_COSINE_DISTANCE` (sub-linear). ANN rules: the `APPROX_*` distance must be the sole `ORDER BY` key and a `LIMIT` is required.

````python
# ANN variant over the whole catalog (uses idx_item_embedding when available).
if ANN_AVAILABLE:
    start = time.perf_counter()
    with database.snapshot() as snapshot:
        ann_candidates = list(snapshot.execute_sql("""
            SELECT item_id, category, price,
                   APPROX_COSINE_DISTANCE(embedding, @user_emb,
                       options => JSON '{"num_leaves_to_search": 100}') AS dist
            FROM item_catalog
            ORDER BY dist
            LIMIT 50
        """, params={'user_emb': user_emb}, param_types={'user_emb': Q_ARRAY_TYPE}))
    ann_ms = (time.perf_counter() - start) * 1000
    print(f'ANN (APPROX_COSINE_DISTANCE, whole catalog) — {len(ann_candidates)} results in {ann_ms:.1f}ms')
    for item_id, cat, price, dist in ann_candidates[:5]:
        print(f'  {item_id}: {cat}, ${price:.2f}, approx_cosine_dist={dist:.4f}')
    print('\n  Tune num_leaves_to_search to trade recall for latency (see Vector Embeddings NB).')
else:
    print('ANN skipped (instance edition lacks VECTOR INDEX support); exact KNN above is the path.')
````

````python
# Stage 2: Feature assembly with JOINs.
# Join the Stage-1 candidates back to user interaction history.
candidate_ids = [c[0] for c in candidates[:20]]  # top-20 by vector distance
placeholders = ', '.join(f"'{cid}'" for cid in candidate_ids)

start = time.perf_counter()
with database.snapshot(exact_staleness=datetime.timedelta(seconds=10), multi_use=True) as snapshot:
    user = list(snapshot.execute_sql(
        "SELECT user_id, age, preferred_category, avg_spend, lifetime_purchases "
        "FROM user_profiles WHERE user_id = @uid",
        params={'uid': user_id}, param_types={'uid': spanner.param_types.STRING},
    ))

    features = list(snapshot.execute_sql(f"""
        SELECT
            i.item_id, i.category, i.price, i.rating, i.popularity_score,
            COALESCE(x.interaction_count, 0) AS prior_interactions,
            x.interaction_type AS last_interaction_type
        FROM item_catalog i
        LEFT JOIN user_item_interactions x
            ON i.item_id = x.item_id AND x.user_id = @uid
        WHERE i.item_id IN ({placeholders})
        ORDER BY i.popularity_score DESC
    """, params={'uid': user_id}, param_types={'uid': spanner.param_types.STRING}))
feature_ms = (time.perf_counter() - start) * 1000

print('Stage 2: Feature assembly with JOINs')
print(f'  User: {user[0][0]} (age={user[0][1]}, pref={user[0][2]}, avg_spend=${user[0][3]:.2f})')
print(f'  Items with interaction history: {len(features)}')
print(f'  Latency: {feature_ms:.1f}ms')
print('\n  Sample features:')
for row in features[:5]:
    print(f'    {row[0]}: {row[1]}, ${row[2]:.2f}, rating={row[3]}, pop={row[4]:.1f}, '
          f'prior_interactions={row[5]}, last_type={row[6]}')
````

````python
# Stage 3: Score candidates with a lightweight linear scorer (no model dependency).
user_row = user[0]
user_spend = user_row[3] or 50.0

scores = []
for item in features:
    # price fit (closeness to avg_spend) + rating + popularity + prior interactions
    price_fit = 1.0 - abs(item[2] - user_spend) / max(user_spend, item[2])
    score = (
        0.30 * price_fit +
        0.25 * (item[3] / 5.0) +
        0.25 * (item[4] / 100.0) +
        0.20 * min(item[5] / 10.0, 1.0)  # prior interactions
    )
    scores.append((item[0], item[1], item[2], item[3], score))

top_10 = sorted(scores, key=lambda x: -x[4])[:10]

print('Stage 3: Scored and ranked (linear scorer)')
print(f'\n  Top 10 recommendations for {user_id}:')
print(f'  {"Item":>15s} {"Category":>12s} {"Price":>8s} {"Rating":>7s} {"Score":>7s}')
for item_id, cat, price, rating, score in top_10:
    print(f'  {item_id:>15s} {cat:>12s} ${price:>6.2f} {rating:>6.1f} {score:>6.3f}')
````

````python
# Stage 4: Atomic update — record the recommendation interaction
def record_recommendation(transaction, user_id, item_id):
    """Atomically update interaction count and item popularity across two tables."""
    # Read current interaction count
    existing = list(transaction.execute_sql(
        "SELECT interaction_count FROM user_item_interactions "
        "WHERE user_id = @uid AND item_id = @iid",
        params={'uid': user_id, 'iid': item_id},
        param_types={'uid': spanner.param_types.STRING, 'iid': spanner.param_types.STRING},
    ))
    new_count = (existing[0][0] + 1) if existing else 1

    # last_interaction is a plain TIMESTAMP (no allow_commit_timestamp) -> use a real time.
    now = datetime.datetime.now(tz=datetime.timezone.utc)

    # Update interaction (updated_at IS a commit-timestamp column)
    transaction.insert_or_update(
        table='user_item_interactions',
        columns=['user_id', 'item_id', 'interaction_type', 'interaction_count',
                 'last_interaction', 'updated_at'],
        values=[[user_id, item_id, 'recommendation', new_count,
                 now, spanner.COMMIT_TIMESTAMP]],
    )

    # Update item popularity score
    pop = list(transaction.execute_sql(
        "SELECT popularity_score FROM item_catalog WHERE item_id = @iid",
        params={'iid': item_id}, param_types={'iid': spanner.param_types.STRING},
    ))
    transaction.update(
        table='item_catalog',
        columns=['item_id', 'popularity_score', 'updated_at'],
        values=[[item_id, pop[0][0] + 0.1, spanner.COMMIT_TIMESTAMP]],
    )

    return new_count

recommended_item = top_10[0][0]
start = time.perf_counter()
new_count = database.run_in_transaction(record_recommendation, user_id, recommended_item)
update_ms = (time.perf_counter() - start) * 1000

print(f'Stage 4: Atomic cross-table update')
print(f'  Updated user_item_interactions: {user_id} × {recommended_item} (count={new_count})')
print(f'  Updated item_catalog: {recommended_item} popularity += 0.1')
print(f'  Latency: {update_ms:.1f}ms')
print(f'  Both tables updated atomically — impossible in Bigtable.')
````

### 3b. End-to-End Pipeline Latency

Wrap Stages 1–3 into a single function and benchmark it across many users to get per-stage median/p95 timings.

````python
def full_recommendation_pipeline(user_id, top_k=5):
    """Full pipeline: read user -> filter+vector KNN -> score. Returns (top, timings)."""
    timings = {}

    # Read user profile (incl. taste vector)
    t0 = time.perf_counter()
    with database.snapshot(exact_staleness=datetime.timedelta(seconds=10)) as snapshot:
        urow = list(snapshot.execute_sql(
            "SELECT preferred_category, avg_spend, embedding FROM user_profiles WHERE user_id = @uid",
            params={'uid': user_id}, param_types={'uid': spanner.param_types.STRING},
        ))
    timings['user_read'] = (time.perf_counter() - t0) * 1000
    if not urow:
        return None, timings

    pref_cat = urow[0][0] or 'electronics'
    avg_spend = urow[0][1] or 50.0
    u_emb = [float(x) for x in urow[0][2]]

    # Stage 1: SQL filter + native vector KNN in one statement
    t1 = time.perf_counter()
    with database.snapshot(exact_staleness=datetime.timedelta(seconds=10)) as snapshot:
        candidates = list(snapshot.execute_sql("""
            SELECT item_id, price, rating, popularity_score,
                   COSINE_DISTANCE(embedding, @q) AS dist
            FROM item_catalog
            WHERE category = @cat AND price < @max_price AND rating > 3.0
            ORDER BY dist
            LIMIT 20
        """, params={'q': u_emb, 'cat': pref_cat, 'max_price': avg_spend * 2.0},
             param_types={'q': Q_ARRAY_TYPE, 'cat': spanner.param_types.STRING,
                          'max_price': spanner.param_types.FLOAT64}))
    timings['candidates'] = (time.perf_counter() - t1) * 1000

    # Stage 2+3: score (closer vector = higher base; blend rating/popularity)
    t2 = time.perf_counter()
    scored = []
    for item_id, price, rating, pop, dist in candidates:
        score = 0.4 * (1.0 - dist) + 0.3 * (rating / 5.0) + 0.3 * (pop / 100.0)
        scored.append((item_id, score))
    top = sorted(scored, key=lambda x: -x[1])[:top_k]
    timings['score'] = (time.perf_counter() - t2) * 1000

    timings['total'] = sum(timings.values())
    return top, timings

# Benchmark
n_runs = 30
all_timings = []
for i in range(n_runs):
    uid = f'user_{(i * 37 % 1000):06d}'
    _, timings = full_recommendation_pipeline(uid)
    all_timings.append(timings)

print(f'End-to-end recommendation pipeline ({n_runs} runs):\n')
print(f'{"Step":<15s} {"Median":>10s} {"p95":>10s}')
for step in ['user_read', 'candidates', 'score', 'total']:
    vals = sorted([t[step] for t in all_timings])
    print(f'{step:<15s} {vals[n_runs//2]:>8.1f}ms {vals[int(n_runs*0.95)]:>8.1f}ms')
````

````python
# Visualize pipeline stages
stages = ['user_read', 'candidates', 'score']
medians = [sorted([t[s] for t in all_timings])[n_runs//2] for s in stages]

fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#4285F4', '#34A853', '#FBBC04']
bars = ax.bar(stages, medians, color=colors, alpha=0.8)

for bar, val in zip(bars, medians):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val:.1f}ms', ha='center', fontsize=11)

ax.set_ylabel('Median Latency (ms)', fontweight='bold')
ax.set_title('Recommendation Pipeline Latency Breakdown', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
````

## 4. Serving Endpoint — FastAPI with Prepared-Statement-Style Queries

Production feature serving runs behind an HTTP API. We build a [FastAPI](https://fastapi.tiangolo.com/) app with:

- A **module-scoped Spanner client** created once at import — the gRPC channel and session pool are reused across requests (the connection-reuse pattern; recreating a client per request is the classic latency bug).
- **Parameterized queries** (`@param` bind variables, never string interpolation). This is the **prepared-statement analog** in Spanner: the query text is fixed and reusable, values bind separately — the plan can be cached server-side and there is no SQL-injection surface.
- A `POST /recommend` endpoint (filter + vector KNN + score) and a `GET /health` endpoint (a trivial `SELECT 1` liveness probe).

We exercise it in-process with `fastapi.testclient.TestClient` (backed by `httpx`) — reliable in a notebook, no separate server process or port to manage.

````python
from fastapi import FastAPI
from fastapi.testclient import TestClient
from pydantic import BaseModel

# --- Module-scoped client (connection reuse) ---
# In a real service this lives at module import; the session pool is shared by all
# requests. Here we reuse the notebook's `database` handle for the same effect.
app = FastAPI(title='Spanner Feature Store — Recommendations')

class RecRequest(BaseModel):
    user_id: str
    top_k: int = 5

@app.get('/health')
def health():
    # Liveness: a trivial parameter-free read confirms the session pool is serving.
    with database.snapshot(multi_use=True) as snapshot:
        list(snapshot.execute_sql('SELECT 1'))
    return {'status': 'ok'}

@app.post('/recommend')
def recommend(req: RecRequest):
    # Prepared-statement analog: fixed SQL text + @bind params (no interpolation).
    with database.snapshot(exact_staleness=datetime.timedelta(seconds=10), multi_use=True) as snapshot:
        urow = list(snapshot.execute_sql(
            'SELECT preferred_category, avg_spend, embedding FROM user_profiles WHERE user_id = @uid',
            params={'uid': req.user_id}, param_types={'uid': spanner.param_types.STRING},
        ))
        if not urow:
            return {'user_id': req.user_id, 'recommendations': [], 'error': 'unknown user'}
        pref_cat, avg_spend = urow[0][0] or 'electronics', urow[0][1] or 50.0
        u_emb = [float(x) for x in urow[0][2]]

        rows = list(snapshot.execute_sql(
            'SELECT item_id, price, rating, popularity_score, '
            'COSINE_DISTANCE(embedding, @q) AS dist '
            'FROM item_catalog '
            'WHERE category = @cat AND price < @max_price AND rating > 3.0 '
            'ORDER BY dist LIMIT 20',
            params={'q': u_emb, 'cat': pref_cat, 'max_price': avg_spend * 2.0},
            param_types={'q': Q_ARRAY_TYPE, 'cat': spanner.param_types.STRING,
                         'max_price': spanner.param_types.FLOAT64}))

    scored = [(r[0], 0.4 * (1.0 - r[4]) + 0.3 * (r[2] / 5.0) + 0.3 * (r[3] / 100.0)) for r in rows]
    top = sorted(scored, key=lambda x: -x[1])[:req.top_k]
    return {'user_id': req.user_id,
            'recommendations': [{'item_id': iid, 'score': round(s, 4)} for iid, s in top]}

client = TestClient(app)

print('GET  /health   ->', client.get('/health').json())
r = client.post('/recommend', json={'user_id': 'user_000001', 'top_k': 5})
print('POST /recommend ->')
for rec in r.json()['recommendations']:
    print(f'   {rec["item_id"]}  score={rec["score"]}')
````

## 5. Concurrent Load Test

The per-stage benchmark above is sequential. Production serving is **concurrent** — many requests in flight at once, contending for the session pool. A `ThreadPoolExecutor` driver hits the `/recommend` endpoint at N concurrent workers and reports **QPS, p50/p95/p99, and error rate** — the numbers an SRE actually pages on.

````python
from concurrent.futures import ThreadPoolExecutor, as_completed

def one_request(uid):
    start = time.perf_counter()
    try:
        resp = client.post('/recommend', json={'user_id': uid, 'top_k': 5})
        ok = resp.status_code == 200 and 'recommendations' in resp.json()
    except Exception:
        ok = False
    return (time.perf_counter() - start) * 1000, ok

N_REQUESTS = 300
uids = [f'user_{(i * 37 % 1000):06d}' for i in range(N_REQUESTS)]

def run_load(workers):
    lats, errors = [], 0
    wall_start = time.perf_counter()
    with ThreadPoolExecutor(max_workers=workers) as ex:
        futures = [ex.submit(one_request, u) for u in uids]
        for f in as_completed(futures):
            ms, ok = f.result()
            lats.append(ms)
            errors += (not ok)
    wall = time.perf_counter() - wall_start
    s = sorted(lats)
    return {
        'workers': workers,
        'qps': N_REQUESTS / wall,
        'p50': s[int(len(s) * 0.50)],
        'p95': s[int(len(s) * 0.95)],
        'p99': s[int(len(s) * 0.99)],
        'err_pct': 100.0 * errors / N_REQUESTS,
    }

print(f'Concurrent load test — {N_REQUESTS} requests through /recommend:')
print()
print(f'{"Workers":>8s} {"QPS":>8s} {"p50":>8s} {"p95":>8s} {"p99":>8s} {"Err%":>7s}')
load_results = []
for w in [1, 8, 32]:
    res = run_load(w)
    load_results.append(res)
    print(f'{res["workers"]:>8d} {res["qps"]:>8.0f} {res["p50"]:>6.1f}ms {res["p95"]:>6.1f}ms {res["p99"]:>6.1f}ms {res["err_pct"]:>6.1f}%')

print()
print('QPS rises with concurrency until the session pool / instance CPU saturates;')
print('p99 climbing while QPS flattens is the signal to add processing units (autoscaling).')
````

````python
# Visualize QPS and tail latency vs concurrency.
ws = [r['workers'] for r in load_results]
qps = [r['qps'] for r in load_results]
p99 = [r['p99'] for r in load_results]

fig, ax1 = plt.subplots(figsize=(9, 5))
ax1.plot(ws, qps, 'o-', color='#34A853', label='QPS')
ax1.set_xlabel('Concurrent workers', fontweight='bold')
ax1.set_ylabel('QPS', color='#34A853', fontweight='bold')
ax1.tick_params(axis='y', labelcolor='#34A853')

ax2 = ax1.twinx()
ax2.plot(ws, p99, 's--', color='#EA4335', label='p99 latency')
ax2.set_ylabel('p99 latency (ms)', color='#EA4335', fontweight='bold')
ax2.tick_params(axis='y', labelcolor='#EA4335')

ax1.set_title('Recommendation Endpoint — QPS vs Tail Latency under Load', fontweight='bold')
ax1.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
````

## 6. Production Operations — Monitoring

A Spanner feature store is observed through **Cloud Monitoring** metrics and **`SPANNER_SYS`** introspection tables. The headline signal is **high-priority CPU utilization**: Spanner's published guidance is to keep a **regional** instance below **65%** high-priority CPU (45% for multi-region). Crossing 65% is the autoscaling trigger and the right place for a high-priority alert.

Key metrics:

| Metric | What it tells you | Alert |
|---|---|---|
| `spanner.googleapis.com/instance/cpu/utilization` (high-priority) | Serving CPU headroom | **> 65%** (regional) |
| `spanner.googleapis.com/api/request_latencies` | Server-side request latency distribution | p99 above SLO |
| `spanner.googleapis.com/instance/storage/used_bytes` | Storage vs the per-node limit | > 95% of limit |

`SPANNER_SYS.QUERY_STATS_TOP_*` (minute / 10-minute / hour windows) surface the most expensive queries — the first place to look when CPU climbs.

````python
# Top queries by execution count over the last 10 minutes (SPANNER_SYS introspection).
# These tables are populated by the live instance; safe to read at any time.
try:
    with database.snapshot() as snapshot:
        top_q = list(snapshot.execute_sql(
            'SELECT text, execution_count, avg_latency_seconds, avg_cpu_seconds '
            'FROM SPANNER_SYS.QUERY_STATS_TOP_10MINUTE '
            'ORDER BY execution_count DESC LIMIT 5'))
    if top_q:
        print('Top queries (last 10 min) by execution count:')
        print()
        for text, n, lat, cpu in top_q:
            snippet = ' '.join(text.split())[:70]
            print(f'  n={n:<6} avg_lat={lat*1000:7.2f}ms avg_cpu={cpu*1000:7.2f}ms  {snippet}')
    else:
        print('No query stats yet (run some queries, then re-read).')
except Exception as e:
    print(f'QUERY_STATS read: {str(e)[:160]}')

print()
print('Cloud Monitoring alert policies to create (gcloud / Terraform / console):')
print('  spanner.googleapis.com/instance/cpu/utilization (high-priority) > 0.65  [regional]')
print('  spanner.googleapis.com/api/request_latencies   p99 > SLO')
print('  spanner.googleapis.com/instance/storage/used_bytes > 95% of limit')
````

## 7. Security — Least-Privilege IAM

The starter config granted a single broad binding. Production splits access into **three least-privilege roles**, matching the Terraform IAM block in Section 1:

| Principal | Role | Why |
|---|---|---|
| Serving SA | `roles/spanner.databaseReader` | Read-only — the serving path never writes |
| Sync SA | `roles/spanner.databaseUser` | Read + DML — runs the BQ → Spanner sync |
| Admin SA | `roles/spanner.databaseAdmin` | Schema + instance management only |

`databaseReader` cannot run DML or change schema; `databaseUser` cannot manage the instance; only `databaseAdmin` can alter schema. A compromised serving credential can read features but cannot mutate or destroy them.

````python
iam_split = f"""
# Least-privilege IAM for the Spanner feature store (gcloud form; Terraform in section 1).

# Serving SA -- read-only (the serving path never writes)
gcloud spanner databases add-iam-policy-binding {SPANNER_DATABASE} \\
    --instance {SPANNER_INSTANCE} \\
    --member serviceAccount:serving-sa@{PROJECT_ID}.iam.gserviceaccount.com \\
    --role roles/spanner.databaseReader

# Sync SA -- read + DML (BQ -> Spanner sync)
gcloud spanner databases add-iam-policy-binding {SPANNER_DATABASE} \\
    --instance {SPANNER_INSTANCE} \\
    --member serviceAccount:sync-sa@{PROJECT_ID}.iam.gserviceaccount.com \\
    --role roles/spanner.databaseUser

# Admin SA -- schema + instance management only
gcloud spanner databases add-iam-policy-binding {SPANNER_DATABASE} \\
    --instance {SPANNER_INSTANCE} \\
    --member serviceAccount:admin-sa@{PROJECT_ID}.iam.gserviceaccount.com \\
    --role roles/spanner.databaseAdmin
"""
print(iam_split)
````

## 8. High Availability and Scaling

**Regional HA (this series).** A regional instance runs **3 read-write replicas across 3 zones** with synchronous Paxos writes and **automatic failover** — a zone failure is transparent, with **zero data loss**. SLA: **99.99%**.

**Multi-region.** A multi-regional config (e.g. `nam3`) survives a full **region** failure and raises the SLA to **99.999%**, at roughly 3× cost. See the [Multi-Region](./Spanner%20Feature%20Store%20-%20Multi-Region.ipynb) notebook for replica types, directed reads, and the stale-read latency trade-off.

**Scaling: processing units, not nodes.** Spanner bills in **processing units** (100 PU = 0.1 node; 1000 PU = 1 node), so you scale in fine increments well below a whole node. **Autoscaling** (set in the Terraform `autoscaling_config`) tracks load between a min/max bound using the 65% high-priority-CPU target — capacity follows demand while the ceiling caps spend. This is the production default; fixed capacity is only for predictable, steady workloads.

````python
# Inspect current capacity. Autoscaling reports processing_units = 0 with an
# autoscaling_config set; fixed-capacity instances report the PU count directly.
instance.reload()
print(f'Instance:         {instance.instance_id}')
print(f'Config:           {instance.configuration_name}')
pu = instance.processing_units
nodes = instance.node_count
if pu:
    print(f'Processing units: {pu}  (~{pu/1000:.1f} node-equivalents)')
else:
    print('Processing units: managed by autoscaling_config')
print(f'Node count:       {nodes}')
print()
print('Regional (this series): 3 read-write replicas / 3 zones, auto failover, 99.99% SLA.')
print('Multi-region (nam3):    survives a region failure, 99.999% SLA (~3x cost).')
print('Scale by processing units (100 PU = 0.1 node); autoscaling tracks the 65% CPU target.')
````

## 9. Production Readiness Checklist

A consolidated checklist for running a Spanner feature store in production — synthesizing the series.

**Infrastructure**
- [ ] `deletion_protection = true` on the instance **and** database (Terraform section 1)
- [ ] `autoscaling_config` with sensible min/max PU bounds (capacity tracks load, ceiling caps spend)
- [ ] Daily `google_spanner_backup_schedule` with a defined retention window
- [ ] Point-in-time recovery: `version_retention_period` set (e.g. `7d`)
- [ ] Reproducible from Terraform (`variables.tf` / `main.tf` / `outputs.tf`); state in a remote backend

**Schema & Data**
- [ ] Embeddings typed `ARRAY<FLOAT32>(vector_length=>N) NOT NULL`; ScaNN `VECTOR INDEX` built for ANN
- [ ] Secondary indexes `STORING` the columns hot queries need (index-only scans)
- [ ] BigQuery is the source of truth; Spanner rebuildable from the sync pipeline

**Serving**
- [ ] Module-scoped client (session pool reused); never one client per request
- [ ] Parameterized queries everywhere (prepared-statement analog; no string interpolation)
- [ ] Stale reads (`exact_staleness`) on the serving path where freshness allows
- [ ] Load tested at expected peak QPS with p99 within SLO (this notebook section 5)

**Operations**
- [ ] Cloud Monitoring alert: high-priority CPU > 65% (regional) / 45% (multi-region)
- [ ] Request-latency p99 and storage-utilization alerts
- [ ] Least-privilege IAM split: serving (`databaseReader`) / sync (`databaseUser`) / admin (`databaseAdmin`)
- [ ] HA matched to SLA need: regional (99.99%) vs multi-region (99.999%)

## 11. Comparison Across Capstones

The Bigtable and Valkey capstones build similar recommendation/serving patterns, each leaning on its store's strengths.

| Aspect | Spanner Capstone | Bigtable Capstone | Valkey Capstone |
|---|---|---|---|
| **Data model** | 3 typed tables + JOINs | 1 denormalized table | hashes + sorted sets |
| **Candidate retrieval** | **SQL `WHERE` + native vector KNN in one query** | KNN on vectors (no SQL filtering) | HNSW ANN + TAG/NUMERIC filter (separate store) |
| **Feature assembly** | JOIN across tables | single-row read (pre-joined) | pipelined hash reads |
| **Atomic update** | multi-table transaction | single-row atomic counter | `HINCRBYFLOAT` |
| **Schema** | typed DDL (price is FLOAT64) | schema-free (bytes) | schema-free |
| **Latency** | ~10–20 ms total | ~5–10 ms total | <5 ms total |

### The vector-search dividing line

| Store | Stage-1 vector retrieval | Metadata filter |
|---|---|---|
| **Valkey** | HNSW ANN, **<5 ms** | TAG/NUMERIC, but in a separate store from relational data |
| **Bigtable** | `COSINE_DISTANCE()` **brute-force, ~100 ms** | row-key prefix only — no rich predicates |
| **Spanner** | **native KNN + SQL filter in one statement** (exact, or ScaNN ANN) | any column, transaction-consistent, same query |

This is the one thing Bigtable's capstone genuinely cannot do: apply a relational predicate *and* a vector ranking in the same step. Valkey can filter + ANN but only over data mirrored into Valkey; Spanner runs both against the system of record.

**Spanner wins when:**
- Filter criteria change (add a `WHERE` clause, not a new table design)
- Features from multiple entity types must be consistent (ACID)
- A vector ranking must be combined with relational predicates in one query

**Bigtable wins when:** sub-5 ms point lookups dominate and the access pattern is fixed.
**Valkey wins when:** sub-ms latency, TTL/expiry, and native ANN are all critical (e.g. trading).
