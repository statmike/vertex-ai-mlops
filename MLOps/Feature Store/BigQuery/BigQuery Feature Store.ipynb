# BigQuery Feature Store

BigQuery is the analytical powerhouse for feature engineering — window functions, joins, ML functions — but standard queries take seconds because every query must schedule a job, scan tables, and return results. The [BigQuery Storage Read API](https://cloud.google.com/bigquery/docs/reference/storage) reads directly from the storage layer — bypassing the query engine entirely — streaming [Apache Arrow](https://arrow.apache.org/) record batches over parallel gRPC connections. Combined with **session caching** and proper **table clustering**, this creates a feature serving layer with 20–200ms latency, zero infrastructure, and capabilities that a key-value store can't match.

**The positioning:** This is a third approach to feature serving on Google Cloud — alongside [Vertex AI Feature Store](../vertex/readme.md) (managed) and [Bigtable Feature Store](../Bigtable/readme.md) (self-managed online store). Rather than trying to beat Bigtable at point lookups (<10ms), this approach eliminates the online store entirely. No sync pipelines, no infrastructure to manage, no key design constraints — while being dramatically faster than standard BigQuery queries and uniquely powerful for bulk reads, flexible filtering, and in-memory vector operations.

**Prerequisites:** Run [BigQuery Feature Store — Environment](./BigQuery%20Feature%20Store%20-%20Environment.ipynb) first to create the dataset and all required tables.

**Reference:** For Storage Read API fundamentals and benchmarks, see [bq-reads](../../../data+ai/tabular-data/bq-reads/bq-reads.ipynb).

**What you'll learn:**
- Why the Storage Read API occupies a latency sweet spot between standard BQ queries (seconds) and Bigtable (<10ms)
- How table clustering and `row_restriction` enable storage-level block pruning for fast filtered reads
- Session caching with a reusable `FeatureSessionPool` that cuts repeat-read latency roughly in half
- Arrow zero-copy deserialization — from gRPC stream to numpy array with no intermediate copies
- A FastAPI serving endpoint with async integration, connection pooling, and concurrency scaling
- Multi-stream parallelism and bulk feature hydration where the Storage API dominates Bigtable at scale
- In-memory vector similarity search with numpy (SIMD-optimized, faster than server-side KNN for moderate scale)
- Honest three-way latency and cost comparisons: standard BQ query vs Storage API vs Bigtable

## 3. Storage Read API — The Core Pattern

The [BigQuery Storage Read API](https://cloud.google.com/bigquery/docs/reference/storage) bypasses the query engine entirely. Instead of submitting a SQL job, it:

1. **Creates a read session** — specifies the table, selected columns, and row filter
2. **Opens one or more streams** — each stream reads a partition of the data
3. **Streams Arrow record batches** — binary data in [Apache Arrow](https://arrow.apache.org/) format over gRPC

The key insight: Arrow's on-wire format matches its in-memory format. There's no deserialization step — the bytes that arrive over gRPC are the bytes your code works with. This is **zero-copy** in the truest sense.

```
Standard BQ Query              Storage Read API
─────────────────              ─────────────────
SQL parse & plan               (skipped)
Slot allocation                (skipped)
Distributed scan               Direct storage read
Result serialization           Arrow zero-copy stream
HTTP response                  gRPC binary stream
Client deserialization         (zero-copy — Arrow IS the format)
─────────────────              ─────────────────
~2–7 seconds                   ~20–200 ms
```

### Helper Function

A reusable function for reading from the Storage Read API. This pattern is used throughout the notebook.

````python
TABLE_PATH = f'projects/{BQ_PROJECT}/datasets/{BQ_DATASET}/tables/{BQ_TABLE_CLUSTERED}'

def storage_read(row_restriction, selected_fields=None, table=TABLE_PATH):
    """Read from BQ Storage API. Returns (arrow_table, session_ms, read_ms)."""
    if selected_fields is None:
        selected_fields = ['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL

    t0 = time.perf_counter()
    session = bqstorage.create_read_session(
        request=dict(
            parent=f'projects/{BQ_PROJECT}',
            read_session=dict(
                table=table,
                data_format=bqs_types.DataFormat.ARROW,
                read_options=dict(
                    selected_fields=selected_fields,
                    row_restriction=row_restriction,
                ),
            ),
            max_stream_count=1,
        )
    )
    session_ms = (time.perf_counter() - t0) * 1000

    t1 = time.perf_counter()
    if session.streams:
        reader = bqstorage.read_rows(session.streams[0].name)
        arrow_table = reader.to_arrow(session)
    else:
        arrow_table = pa.table({})
    read_ms = (time.perf_counter() - t1) * 1000

    return arrow_table, session_ms, read_ms
````

### Single-Entity Read

````python
# Single entity read via Storage Read API
restriction = f"entity_group = '{TEST_GROUP}' AND entity_id = '{TEST_ID}'"

arrow_table, session_ms, read_ms = storage_read(restriction)
total_ms = session_ms + read_ms

print(f'Storage Read API — single entity:')
print(f'  Session creation: {session_ms:>7.1f} ms')
print(f'  Stream read:      {read_ms:>7.1f} ms')
print(f'  Total:            {total_ms:>7.1f} ms')
print(f'  Rows:             {arrow_table.num_rows}')
print(f'  Columns:          {arrow_table.num_columns}')
print(f'\nSpeedup vs BQ query: {bq_single_ms / total_ms:.0f}x')
print(f'\nData:')
print(arrow_table.to_pandas().to_string(index=False))
````

### Latency Breakdown

Session creation is the dominant cost — it's where BigQuery resolves the table metadata, applies the `row_restriction` to identify relevant storage blocks, and allocates streams. The actual data read over gRPC is fast because Arrow streaming is zero-copy.

````python
# Run multiple iterations to measure variability
N_ITER = 10
session_times = []
read_times = []

for i in range(N_ITER):
    group = chr(ord('A') + i % 26)
    entity = str((i * 137 + 1) % 5000 + 1).zfill(6)
    restriction = f"entity_group = '{group}' AND entity_id = '{entity}'"

    _, s_ms, r_ms = storage_read(restriction)
    session_times.append(s_ms)
    read_times.append(r_ms)
    print(f'  Iter {i+1:2d}: session={s_ms:>7.1f}ms  read={r_ms:>6.1f}ms  total={s_ms+r_ms:>7.1f}ms')

session_median = sorted(session_times)[N_ITER // 2]
read_median = sorted(read_times)[N_ITER // 2]
total_median = session_median + read_median

print(f'\nMedian latency:')
print(f'  Session creation: {session_median:>7.1f} ms ({session_median/total_median*100:.0f}% of total)')
print(f'  Stream read:      {read_median:>7.1f} ms ({read_median/total_median*100:.0f}% of total)')
print(f'  Total:            {total_median:>7.1f} ms')
````

### Batch Reads

````python
# Batch reads: 10, 100, 1K, 10K entities
random.seed(42)
batch_sizes_api = [10, 100, 1000, 10000]
api_batch_results = {}

for n in batch_sizes_api:
    ids = [(random.choice('ABCDEFGHIJKLMNOPQRSTUVWXYZ'),
            str(random.randint(1, 5000)).zfill(6)) for _ in range(n)]
    # Use entity_group IN + entity_id IN for clustered reads
    id_pairs = ', '.join(f"('{g}', '{e}')" for g, e in ids)
    restriction = f"(entity_group, entity_id) IN ({id_pairs})"

    start = time.perf_counter()
    session = bqstorage.create_read_session(
        request=dict(
            parent=f'projects/{BQ_PROJECT}',
            read_session=dict(
                table=TABLE_PATH,
                data_format=bqs_types.DataFormat.ARROW,
                read_options=dict(
                    selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL,
                    row_restriction=restriction,
                ),
            ),
            max_stream_count=4,
        )
    )

    tables = []
    for stream in session.streams:
        reader = bqstorage.read_rows(stream.name)
        tables.append(reader.to_arrow(session))

    if tables:
        combined = pa.concat_tables(tables)
    else:
        combined = pa.table({})

    latency_ms = (time.perf_counter() - start) * 1000
    api_batch_results[n] = latency_ms

    print(f'Batch {n:>6,d} entities: {latency_ms:>8,.0f} ms  '
          f'({combined.num_rows} rows, {len(session.streams)} streams)')

print(f'\nComparison to standard BQ query (same batch sizes):')
for n in [10, 100, 1000]:
    if n in bq_batch_results and n in api_batch_results:
        speedup = bq_batch_results[n] / api_batch_results[n]
        print(f'  {n:>5,d} entities: BQ={bq_batch_results[n]:>6,.0f}ms  '
              f'Storage API={api_batch_results[n]:>6,.0f}ms  ({speedup:.0f}x faster)')
````

### Column Pruning

BigQuery uses [columnar storage](https://cloud.google.com/bigquery/docs/storage_overview). When you specify `selected_fields`, only the requested columns are read from disk — the rest are never touched. This is fundamentally different from a row-oriented store like Bigtable, where reading any column requires reading the entire column family.

````python
# Column pruning: read 5 vs 50 vs all float columns
restriction = f"entity_group = '{TEST_GROUP}'"

column_sets = {
    '5 columns': ['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL,
    '50 columns': ['entity_group', 'entity_id'] + FEATURE_COLUMNS_MEDIUM,
    'All columns': None,  # None = all columns
}

for label, cols in column_sets.items():
    start = time.perf_counter()
    if cols is None:
        session = bqstorage.create_read_session(
            request=dict(
                parent=f'projects/{BQ_PROJECT}',
                read_session=dict(
                    table=TABLE_PATH,
                    data_format=bqs_types.DataFormat.ARROW,
                    read_options=dict(row_restriction=restriction),
                ),
                max_stream_count=1,
            )
        )
    else:
        session = bqstorage.create_read_session(
            request=dict(
                parent=f'projects/{BQ_PROJECT}',
                read_session=dict(
                    table=TABLE_PATH,
                    data_format=bqs_types.DataFormat.ARROW,
                    read_options=dict(
                        selected_fields=cols,
                        row_restriction=restriction,
                    ),
                ),
                max_stream_count=1,
            )
        )

    if session.streams:
        reader = bqstorage.read_rows(session.streams[0].name)
        at = reader.to_arrow(session)
    else:
        at = pa.table({})

    latency_ms = (time.perf_counter() - start) * 1000
    n_cols = at.num_columns if at.num_columns > 0 else (len(cols) if cols else 'all')
    bytes_read = at.nbytes
    print(f'{label:<15s}: {latency_ms:>7.0f} ms  ({at.num_rows} rows, {n_cols} cols, {bytes_read/1024:.0f} KB)')

print(f'\nColumn pruning is nearly free — columnar storage skips unread columns entirely.')
````

### Flexible Row Filtering

The `row_restriction` parameter accepts any valid BigQuery [WHERE clause](https://cloud.google.com/bigquery/docs/reference/standard-sql/query-syntax#where_clause). This is where the Storage Read API fundamentally differs from a key-value store: you can filter on **any column, any combination** — no pre-designed row keys required.

````python
# Flexible filtering — queries impossible with Bigtable row keys
filters = {
    'By group': "entity_group = 'A'",
    'Numeric range': "entity_group = 'A' AND feature_float_1 > 50.0",
    'Compound': "entity_group IN ('A', 'B', 'C') AND feature_float_1 > 50.0 AND feature_float_5 < 0.5",
    'String match': "entity_group = 'A' AND entity_id >= '002000' AND entity_id < '003000'",
}

for label, restriction in filters.items():
    at, s_ms, r_ms = storage_read(restriction)
    total = s_ms + r_ms
    print(f'{label:<15s}: {total:>7.0f} ms  ({at.num_rows} rows)')
    print(f'                  restriction: {restriction}')

print(f'\nWith Bigtable, these queries would require either:')
print(f'  - Pre-designed row keys (inflexible, one access pattern per key design)')
print(f'  - Full table scan (defeats the purpose of a key-value store)')
print(f'  - Multiple secondary indexes (complex, expensive to maintain)')
````

### Arrow Zero-Copy Path

The Storage Read API streams data in [Apache Arrow](https://arrow.apache.org/) format. Arrow's in-memory layout is columnar and directly mapped from the wire format — no parsing, no conversion, no copies. Going from Arrow to numpy is a pointer cast, not a data copy.

````python
# Demonstrate the Arrow zero-copy path
restriction = f"entity_group = '{TEST_GROUP}'"
at, _, _ = storage_read(
    restriction,
    selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL,
)

# Path 1: Arrow → numpy (zero-copy for numeric columns)
t0 = time.perf_counter()
np_array = at.column('feature_float_1').to_numpy(zero_copy_only=False)
arrow_to_numpy_us = (time.perf_counter() - t0) * 1e6

# Path 2: Arrow → pandas DataFrame
t0 = time.perf_counter()
df = at.to_pandas()
arrow_to_pandas_us = (time.perf_counter() - t0) * 1e6

# Path 3: Arrow → Python dicts
t0 = time.perf_counter()
dicts = at.to_pydict()
arrow_to_dict_us = (time.perf_counter() - t0) * 1e6

print(f'Arrow conversion times ({at.num_rows} rows):')
print(f'  Arrow → numpy:  {arrow_to_numpy_us:>10,.0f} \u03bcs  (zero-copy for float columns)')
print(f'  Arrow → pandas: {arrow_to_pandas_us:>10,.0f} \u03bcs  (memory copy)')
print(f'  Arrow → dict:   {arrow_to_dict_us:>10,.0f} \u03bcs  (Python objects)')
print(f'\nArrow table memory: {at.nbytes / 1024:.1f} KB')
print(f'numpy array shape:  {np_array.shape}, dtype: {np_array.dtype}')
````

## 4. Session Caching — The Core Optimization

Session creation is the dominant latency cost — typically 50–100ms of an 80–150ms total read. Each `create_read_session()` call negotiates a new session with BigQuery's storage layer: stream allocation, schema resolution, predicate pushdown. For a clustered table with a `row_restriction`, the result is a session with 1–2 streams pointing at the exact storage blocks. That negotiation is pure overhead if you're reading from the same table repeatedly.

**The solution:** Pre-create sessions, cache the session metadata, and reuse streams until they expire.

### FeatureSessionPool

````python
class FeatureSessionPool:
    """Pre-creates and caches BQ Storage Read sessions for fast feature reads."""

    def __init__(self, client, table, project, selected_fields, ttl_seconds=300):
        self.client = client
        self.table = table
        self.project = project
        self.selected_fields = selected_fields
        self.ttl_seconds = ttl_seconds
        self._cache = {}  # key: row_restriction → (session, created_at)

    def get_session(self, row_restriction):
        """Get or create a session for this restriction."""
        self._evict_expired()
        if row_restriction in self._cache:
            return self._cache[row_restriction][0]
        session = self._create(row_restriction)
        self._cache[row_restriction] = (session, time.time())
        return session

    def warm(self, restrictions):
        """Pre-create sessions for known query patterns."""
        for r in restrictions:
            self.get_session(r)

    def read(self, row_restriction):
        """Full read: get session, read stream, return Arrow table."""
        session = self.get_session(row_restriction)
        if not session.streams:
            return pa.table({})
        reader = self.client.read_rows(session.streams[0].name)
        return reader.to_arrow(session)

    def _create(self, row_restriction):
        return self.client.create_read_session(
            request=dict(
                parent=f'projects/{self.project}',
                read_session=dict(
                    table=self.table,
                    data_format=bqs_types.DataFormat.ARROW,
                    read_options=dict(
                        selected_fields=self.selected_fields,
                        row_restriction=row_restriction,
                    ),
                ),
                max_stream_count=1,
            )
        )

    def _evict_expired(self):
        now = time.time()
        expired = [k for k, (_, t) in self._cache.items() if now - t > self.ttl_seconds]
        for k in expired:
            del self._cache[k]

    @property
    def cache_size(self):
        return len(self._cache)

print(f'FeatureSessionPool defined — {48} lines, no external dependencies.')
print(f'Features: TTL-based eviction, warm-up, Arrow reads.')
````

### Session Caching Benchmark

Compare reads with and without session caching to quantify the speedup.

````python
# Initialize pool
pool = FeatureSessionPool(
    client=bqstorage,
    table=TABLE_PATH,
    project=BQ_PROJECT,
    selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL,
    ttl_seconds=300,
)

# Cold read (no cache)
restriction = f"entity_group = '{TEST_GROUP}' AND entity_id = '{TEST_ID}'"

cold_times = []
for i in range(5):
    pool._cache.clear()
    start = time.perf_counter()
    result = pool.read(restriction)
    cold_times.append((time.perf_counter() - start) * 1000)

# Warm read (session cached)
warm_times = []
pool._cache.clear()
pool.warm([restriction])  # Pre-create the session
for i in range(5):
    start = time.perf_counter()
    result = pool.read(restriction)
    warm_times.append((time.perf_counter() - start) * 1000)

cold_median = sorted(cold_times)[len(cold_times) // 2]
warm_median = sorted(warm_times)[len(warm_times) // 2]

print(f'Session caching benchmark:')
print(f'  Cold (no cache):  {cold_median:>7.1f} ms (median of {len(cold_times)} runs)')
print(f'  Warm (cached):    {warm_median:>7.1f} ms (median of {len(warm_times)} runs)')
print(f'  Speedup:          {cold_median / warm_median:.1f}x')
print(f'  Savings:          {cold_median - warm_median:.0f} ms per read')
print(f'\nCache size: {pool.cache_size} sessions')
````

## 5. FastAPI Serving Application

Build a feature serving endpoint using the session pool from Part 4. The architecture:
- `BigQueryReadClient()` initialized **once** at startup — gRPC channel pool reused across all requests
- `FeatureSessionPool` for warm reads
- `asyncio.to_thread()` for non-blocking FastAPI integration
- Input validation to prevent SQL injection in `row_restriction`

> **Note:** Cloud Run deployment, scaling, authentication, and production hardening are covered in a companion notebook: `BigQuery Feature Store - Cloud Run.ipynb`.

````python
import re
import asyncio
from fastapi import FastAPI, HTTPException
from fastapi.responses import JSONResponse

app = FastAPI(title='BigQuery Feature Store')

# Initialized ONCE at startup — reused across all requests
serving_client = bigquery_storage.BigQueryReadClient()
serving_pool = FeatureSessionPool(
    client=serving_client,
    table=TABLE_PATH,
    project=BQ_PROJECT,
    selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL,
    ttl_seconds=300,
)

ENTITY_PATTERN = re.compile(r'^[A-Z]$')
ID_PATTERN = re.compile(r'^\d{6}$')

def read_features(entity_group: str, entity_id: str) -> dict:
    """Read features for a single entity. Runs in a thread pool."""
    restriction = f"entity_group = '{entity_group}' AND entity_id = '{entity_id}'"
    session = serving_pool.get_session(restriction)
    if not session.streams:
        return {}
    reader = serving_client.read_rows(session.streams[0].name)
    arrow_table = reader.to_arrow(session)
    if arrow_table.num_rows == 0:
        return {}
    return {col: arrow_table.column(col).to_pylist()[0] for col in arrow_table.column_names}

@app.get('/features/{entity_group}/{entity_id}')
async def get_features(entity_group: str, entity_id: str):
    if not ENTITY_PATTERN.match(entity_group):
        raise HTTPException(400, 'entity_group must be a single uppercase letter')
    if not ID_PATTERN.match(entity_id):
        raise HTTPException(400, 'entity_id must be a 6-digit zero-padded number')
    features = await asyncio.to_thread(read_features, entity_group, entity_id)
    if not features:
        raise HTTPException(404, 'Entity not found')
    return features

@app.get('/health')
async def health():
    return {'status': 'ok', 'cached_sessions': serving_pool.cache_size}

print(f'FastAPI app defined:')
print(f'  GET /features/{{entity_group}}/{{entity_id}}')
print(f'  GET /health')
print(f'\nInput validation prevents SQL injection in row_restriction.')
````

### Async Reads

The BigQuery Storage Read API client uses synchronous gRPC under the hood. For async serving frameworks like FastAPI, `asyncio.to_thread()` offloads reads to a thread pool — keeping the event loop free for other requests. Multiple reads execute concurrently across threads.

````python
# Async read pattern: sequential vs concurrent
import asyncio

async def async_storage_read(entity_group, entity_id):
    """Non-blocking wrapper around the synchronous Storage Read API."""
    restriction = f"entity_group = '{entity_group}' AND entity_id = '{entity_id}'"
    return await asyncio.to_thread(storage_read, restriction)

# Generate 10 random entities
random.seed(99)
test_entities = [(random.choice('ABCDEFGHIJKLMNOPQRSTUVWXYZ'),
                  str(random.randint(1, 5000)).zfill(6)) for _ in range(10)]

# Sequential reads
start = time.perf_counter()
for g, e in test_entities:
    storage_read(f"entity_group = '{g}' AND entity_id = '{e}'")
sequential_ms = (time.perf_counter() - start) * 1000

# Concurrent reads via asyncio.gather
async def batch_read():
    tasks = [async_storage_read(g, e) for g, e in test_entities]
    return await asyncio.gather(*tasks)

start = time.perf_counter()
results = await batch_read()
concurrent_ms = (time.perf_counter() - start) * 1000

print(f'10 entity reads — sequential vs concurrent:')
print(f'  Sequential:  {sequential_ms:>7.0f} ms  ({sequential_ms/10:.0f} ms/read)')
print(f'  Concurrent:  {concurrent_ms:>7.0f} ms  ({concurrent_ms/10:.0f} ms/read)')
print(f'  Speedup:     {sequential_ms / concurrent_ms:.1f}x')
print(f'\nAll {len(results)} reads returned data — {sum(r[0].num_rows for r in results)} total rows.')
````

````python
# Read an entire entity group (5,000 entities) with parallel streams
restriction = "entity_group = 'A'"

start = time.perf_counter()
session = bqstorage.create_read_session(
    request=dict(
        parent=f'projects/{BQ_PROJECT}',
        read_session=dict(
            table=TABLE_PATH,
            data_format=bqs_types.DataFormat.ARROW,
            read_options=dict(
                selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL,
                row_restriction=restriction,
            ),
        ),
        max_stream_count=4,
    )
)

def read_stream(stream_name):
    reader = bqstorage.read_rows(stream_name)
    return reader.to_arrow(session)

with ThreadPoolExecutor(max_workers=4) as executor:
    tables = list(executor.map(read_stream, [s.name for s in session.streams]))

combined = pa.concat_tables(tables) if tables else pa.table({})
segment_ms = (time.perf_counter() - start) * 1000

print(f'Segment read (group A):')
print(f'  Latency:  {segment_ms:>7.0f} ms')
print(f'  Rows:     {combined.num_rows:,}')
print(f'  Streams:  {len(session.streams)}')
print(f'  Size:     {combined.nbytes / 1024:.0f} KB')
````

### Multi-Stream Scaling

The `max_stream_count` parameter controls how many parallel streams BigQuery allocates for a read session. Each stream reads a disjoint partition of the data. For large reads, parallel streams reduce wall-clock time — but `max_stream_count` is a **hint**, not a guarantee. BigQuery allocates streams based on the **data volume in bytes** (not entity count), so narrow reads (few columns) may receive fewer streams than requested even with many rows.

````python
# Multi-stream scaling: vary max_stream_count for full table read
stream_counts = [1, 2, 4, 8]
stream_results = []

for max_streams in stream_counts:
    start = time.perf_counter()
    session = bqstorage.create_read_session(
        request=dict(
            parent=f'projects/{BQ_PROJECT}',
            read_session=dict(
                table=TABLE_PATH,
                data_format=bqs_types.DataFormat.ARROW,
                read_options=dict(
                    selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL,
                ),
            ),
            max_stream_count=max_streams,
        )
    )

    actual_streams = len(session.streams)

    def read_stream(stream_name):
        reader = bqstorage.read_rows(stream_name)
        return reader.to_arrow(session)

    with ThreadPoolExecutor(max_workers=max_streams) as executor:
        tables = list(executor.map(read_stream, [s.name for s in session.streams]))

    combined = pa.concat_tables(tables) if tables else pa.table({})
    latency_ms = (time.perf_counter() - start) * 1000
    throughput = combined.num_rows / (latency_ms / 1000)

    stream_results.append((max_streams, actual_streams, latency_ms, combined.num_rows, throughput))
    print(f'  Requested={max_streams}  Actual={actual_streams}  '
          f'Latency={latency_ms:>7.0f}ms  Rows={combined.num_rows:>7,}  '
          f'Throughput={throughput:>9,.0f} rows/s')

print(f'\nmax_stream_count is a hint — BigQuery allocates streams based on data volume (bytes),')
print(f'not entity count. With {len(FEATURE_COLUMNS_SMALL) + 2} selected columns (~5 MB total),')
print(f'the data fits in a single stream. Multi-stream benefits appear with wider reads')
print(f'(more columns) or larger tables where there is enough data to partition.')
````

````python
# Measure read time at different scales — Storage API vs standard BQ query
scale_results = []

scale_configs = [
    ('1 entity', f"entity_group = 'A' AND entity_id = '000001'", 1),
    ('10 entities', f"entity_group = 'A' AND entity_id >= '000001' AND entity_id <= '000010'", 1),
    ('100 entities', f"entity_group = 'A' AND entity_id >= '000001' AND entity_id <= '000100'", 1),
    ('1K entities', f"entity_group = 'A' AND entity_id >= '000001' AND entity_id <= '001000'", 2),
    ('5K entities (1 group)', f"entity_group = 'A'", 4),
    ('10K entities (2 groups)', f"entity_group IN ('A', 'B')", 4),
    ('130K entities (all)', None, 8),
]

for label, restriction, max_streams in scale_configs:
    # --- Storage Read API ---
    start = time.perf_counter()

    req = dict(
        parent=f'projects/{BQ_PROJECT}',
        read_session=dict(
            table=TABLE_PATH,
            data_format=bqs_types.DataFormat.ARROW,
            read_options=dict(
                selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS_SMALL,
            ),
        ),
        max_stream_count=max_streams,
    )
    if restriction:
        req['read_session']['read_options']['row_restriction'] = restriction

    session = bqstorage.create_read_session(request=req)

    with ThreadPoolExecutor(max_workers=max_streams) as executor:
        tables = list(executor.map(
            lambda s: bqstorage.read_rows(s).to_arrow(session),
            [s.name for s in session.streams]
        ))

    combined = pa.concat_tables(tables) if tables else pa.table({})
    api_ms = (time.perf_counter() - start) * 1000

    # --- Standard BQ Query ---
    where = f'WHERE {restriction}' if restriction else ''
    bq_sql = f'''
        SELECT entity_group, entity_id, {', '.join(FEATURE_COLUMNS_SMALL)}
        FROM `{BQ_PROJECT}.{BQ_DATASET}.{BQ_TABLE_CLUSTERED}`
        {where}
    '''
    start = time.perf_counter()
    bq.query(bq_sql).to_dataframe()
    bq_ms = (time.perf_counter() - start) * 1000

    scale_results.append((label, combined.num_rows, api_ms, bq_ms, len(session.streams)))
    print(f'{label:<25s}: API={api_ms:>7,.0f}ms  BQ={bq_ms:>7,.0f}ms  '
          f'({combined.num_rows:>7,} rows, {len(session.streams)} streams)')

print(f'\nStorage API latency is nearly flat — it scales with data volume, not entity count.')
print(f'BQ queries carry fixed job-scheduling overhead regardless of result size.')
````

````python
# Visualize scaling: Storage API vs BQ Query
labels = [r[0] for r in scale_results]
api_latencies = [r[2] for r in scale_results]
bq_latencies = [r[3] for r in scale_results]

fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(len(labels))
width = 0.35

bars_api = ax.bar(x - width/2, api_latencies, width, label='Storage Read API', color='#1565C0')
bars_bq = ax.bar(x + width/2, bq_latencies, width, label='Standard BQ Query', color='#E53935')

for bar in bars_api:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(bq_latencies) * 0.01,
            f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8, fontweight='bold', color='#1565C0')
for bar in bars_bq:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(bq_latencies) * 0.01,
            f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8, fontweight='bold', color='#E53935')

ax.set_ylabel('Latency (ms)', fontweight='bold')
ax.set_title('Feature Read Latency — Storage API vs Standard BQ Query', fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=30, ha='right')
ax.legend()
plt.tight_layout()
plt.show()

avg_bq = sum(bq_latencies) / len(bq_latencies)
avg_api = sum(api_latencies) / len(api_latencies)
print(f'Average speedup: {avg_bq / avg_api:.1f}x — Storage API bypasses query engine overhead entirely.')
````

### Training Data Extraction

The zero-copy pipeline: Storage Read API → Arrow → numpy → model. No intermediate DataFrames, no serialization overhead.

````python
# Arrow → numpy zero-copy pipeline for ML training
restriction = "entity_group IN ('A', 'B', 'C', 'D', 'E')"
feature_cols = FEATURE_COLUMNS_SMALL

start = time.perf_counter()
session = bqstorage.create_read_session(
    request=dict(
        parent=f'projects/{BQ_PROJECT}',
        read_session=dict(
            table=TABLE_PATH,
            data_format=bqs_types.DataFormat.ARROW,
            read_options=dict(
                selected_fields=feature_cols + ['label'],
                row_restriction=restriction,
            ),
        ),
        max_stream_count=4,
    )
)

with ThreadPoolExecutor(max_workers=4) as executor:
    tables = list(executor.map(
        lambda s: bqstorage.read_rows(s).to_arrow(session),
        [s.name for s in session.streams]
    ))

combined = pa.concat_tables(tables)
read_ms = (time.perf_counter() - start) * 1000

# Arrow → numpy (zero-copy for float columns)
t0 = time.perf_counter()
X = np.column_stack([combined.column(col).to_numpy(zero_copy_only=False) for col in feature_cols])
y = combined.column('label').to_numpy(zero_copy_only=False)
convert_ms = (time.perf_counter() - t0) * 1000

print(f'Training data extraction:')
print(f'  Read:       {read_ms:>7.0f} ms ({combined.num_rows:,} rows)')
print(f'  Convert:    {convert_ms:>7.1f} ms (Arrow → numpy)')
print(f'  Total:      {read_ms + convert_ms:>7.0f} ms')
print(f'\n  X shape: {X.shape}, dtype: {X.dtype}')
print(f'  y shape: {y.shape}, dtype: {y.dtype}')
print(f'\nReady for model.fit(X, y) — zero-copy from BigQuery storage to numpy arrays.')
````

## 7. In-Memory Vector Similarity

Store embeddings in BigQuery, retrieve via Storage Read API, compute similarity locally with numpy. For moderate-scale vectors (<100K), client-side numpy operations are faster than server-side KNN because numpy leverages SIMD (Single Instruction, Multiple Data) CPU instructions for vectorized math.

Compare to the [Bigtable approach](../Bigtable/Bigtable%20Feature%20Store%20-%20Vector%20Storage%20and%20KNN%20Search.ipynb): Bigtable uses GoogleSQL `COSINE_DISTANCE()` for brute-force KNN on the server, which works but performs a linear scan per query.

````python
# Read all embeddings for a single group (candidates)
EMBEDDINGS_PATH = f'projects/{BQ_PROJECT}/datasets/{BQ_DATASET}/tables/{BQ_TABLE_EMBEDDINGS}'
CANDIDATE_GROUP = 'A'

start = time.perf_counter()
session = bqstorage.create_read_session(
    request=dict(
        parent=f'projects/{BQ_PROJECT}',
        read_session=dict(
            table=EMBEDDINGS_PATH,
            data_format=bqs_types.DataFormat.ARROW,
            read_options=dict(
                selected_fields=['entity_group', 'entity_id', 'embedding'],
                row_restriction=f"entity_group = '{CANDIDATE_GROUP}'",
            ),
        ),
        max_stream_count=1,
    )
)

reader = bqstorage.read_rows(session.streams[0].name)
emb_table = reader.to_arrow(session)
load_ms = (time.perf_counter() - start) * 1000

# Convert embeddings to numpy matrix
t0 = time.perf_counter()
embedding_lists = emb_table.column('embedding').to_pylist()
candidate_matrix = np.array(embedding_lists, dtype=np.float64)
entity_ids = emb_table.column('entity_id').to_pylist()
convert_ms = (time.perf_counter() - t0) * 1000

print(f'Candidate embeddings loaded:')
print(f'  Load:     {load_ms:>7.0f} ms ({emb_table.num_rows} vectors)')
print(f'  Convert:  {convert_ms:>7.1f} ms (Arrow → numpy matrix)')
print(f'  Shape:    {candidate_matrix.shape} (entities x dimensions)')
````

````python
from scipy.spatial.distance import cdist

# Simulate a query: find top-K most similar items to a "user" embedding
query_embedding = candidate_matrix[0]  # Use first entity as query
K = 10

# Method 1: numpy dot product (cosine similarity)
t0 = time.perf_counter()
norms = np.linalg.norm(candidate_matrix, axis=1)
query_norm = np.linalg.norm(query_embedding)
cosine_sim = np.dot(candidate_matrix, query_embedding) / (norms * query_norm + 1e-10)
top_k_indices = np.argsort(cosine_sim)[-K:][::-1]
numpy_us = (time.perf_counter() - t0) * 1e6

# Method 2: scipy cdist
t0 = time.perf_counter()
distances = cdist([query_embedding], candidate_matrix, metric='cosine')[0]
top_k_scipy = np.argsort(distances)[:K]
scipy_us = (time.perf_counter() - t0) * 1e6

print(f'Top-{K} similarity search ({candidate_matrix.shape[0]} candidates, {candidate_matrix.shape[1]} dims):')
print(f'  numpy dot product: {numpy_us:>8,.0f} \u03bcs')
print(f'  scipy cdist:       {scipy_us:>8,.0f} \u03bcs')
print(f'\nTop-{K} results:')
for rank, idx in enumerate(top_k_indices):
    print(f'  {rank+1:2d}. entity_id={entity_ids[idx]}  similarity={cosine_sim[idx]:.4f}')

print(f'\nTotal pipeline: {load_ms:.0f}ms (load) + {numpy_us/1000:.1f}ms (search) = {load_ms + numpy_us/1000:.0f}ms')
print(f'\nCompare to Bigtable KNN: GoogleSQL COSINE_DISTANCE() performs a linear scan')
print(f'on the server. For {candidate_matrix.shape[0]} vectors, numpy SIMD is faster because')
print(f'the computation is vectorized at the CPU instruction level.')
````

## 8. Advanced Patterns

Demonstrate flexibility that's impossible or impractical with a key-value store.

### Multi-Table Reads

Read user features from one table and item embeddings from another, then join in-memory. With Bigtable, each table has its own schema and key design — cross-table joins require application-level logic. Here, it's just two Storage API reads.

````python
# Multi-table read: features + embeddings
entity = "entity_group = 'A' AND entity_id = '000001'"

# Read 1: scalar features
t0 = time.perf_counter()
features_at, _, _ = storage_read(entity)
features_ms = (time.perf_counter() - t0) * 1000

# Read 2: embedding vector
t0 = time.perf_counter()
emb_at, _, _ = storage_read(
    entity,
    selected_fields=['entity_group', 'entity_id', 'embedding'],
    table=EMBEDDINGS_PATH,
)
emb_ms = (time.perf_counter() - t0) * 1000

print(f'Multi-table read for entity A#000001:')
print(f'  Scalar features:  {features_ms:>7.0f} ms ({features_at.num_columns} columns)')
print(f'  Embedding vector: {emb_ms:>7.0f} ms ({len(emb_at.column("embedding").to_pylist()[0])} dims)')
print(f'  Total:            {features_ms + emb_ms:>7.0f} ms')
print(f'\nNo pre-join needed — read from any table, combine in-memory.')
````

### Dynamic Column Selection

Different models need different feature subsets. With columnar storage, changing `selected_fields` is essentially free — only the requested columns are read from disk.

````python
# Different feature subsets for different models
restriction = "entity_group = 'A'"

model_configs = {
    'Model A (5 features)': FEATURE_COLUMNS_SMALL,
    'Model B (50 features)': FEATURE_COLUMNS_MEDIUM,
}

for model_name, cols in model_configs.items():
    at, s_ms, r_ms = storage_read(
        restriction,
        selected_fields=['entity_group', 'entity_id'] + cols,
    )
    total = s_ms + r_ms
    print(f'{model_name}: {total:>7.0f} ms  ({at.num_rows} rows, {at.num_columns} cols, {at.nbytes/1024:.0f} KB)')

print(f'\nSame table, different column subsets — no schema changes, no key redesign.')
````

### Ad-Hoc Filtering

Predicates that would be impossible with Bigtable unless you designed the row key for that exact query pattern.

````python
# Ad-hoc filters impossible with Bigtable row keys
ad_hoc_filters = [
    ('High feature_float_1', "feature_float_1 > 90.0"),
    ('Low feature_float_5', "feature_float_5 < 0.1"),
    ('Compound predicate', "entity_group IN ('A','B','C') AND feature_float_1 > 50 AND feature_float_5 < 0.3"),
    ('Range scan on entity_id', "entity_group = 'A' AND entity_id BETWEEN '002000' AND '003000'"),
]

for label, restriction in ad_hoc_filters:
    at, s_ms, r_ms = storage_read(restriction)
    print(f'{label:<25s}: {s_ms+r_ms:>7.0f} ms  ({at.num_rows:>5,} rows)')
````

### Point-in-Time Reads with BigQuery Time Travel

BigQuery [time travel](https://cloud.google.com/bigquery/docs/time-travel) lets you query data as it existed at a specific timestamp — up to 7 days in the past. No cell versioning, no key-based history — it's built into the storage layer.

````python
# Point-in-time read via BigQuery time travel
# Note: Time travel works via SQL queries, not the Storage Read API directly.
# But the pattern is valuable for training data snapshots.

from datetime import datetime, timedelta, timezone

one_minute_ago = (datetime.now(timezone.utc) - timedelta(minutes=1)).strftime('%Y-%m-%d %H:%M:%S UTC')

time_travel_sql = f'''
    SELECT entity_group, entity_id, feature_float_1
    FROM `{BQ_PROJECT}.{BQ_DATASET}.{BQ_TABLE_DENSE}`
    FOR SYSTEM_TIME AS OF TIMESTAMP '{one_minute_ago}'
    WHERE entity_group = '{TEST_GROUP}' AND entity_id = '{TEST_ID}'
'''

start = time.perf_counter()
try:
    result = bq.query(time_travel_sql).to_dataframe()
    latency_ms = (time.perf_counter() - start) * 1000
    print(f'Time travel query ({one_minute_ago}):')
    print(f'  Latency: {latency_ms:>7.0f} ms')
    print(f'  Result:  {result.to_dict("records")}')
except Exception as e:
    print(f'Time travel query: {e}')
    print(f'  (Table may be too new for time travel)')

print(f'\nTime travel supports lookback up to 7 days — enabling reproducible')
print(f'feature snapshots for training without explicit versioning.')
````
