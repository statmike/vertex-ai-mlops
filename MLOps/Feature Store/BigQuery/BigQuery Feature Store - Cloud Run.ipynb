# BigQuery Feature Store — Cloud Run

Deploy the BigQuery feature serving app from the [main notebook](./BigQuery%20Feature%20Store.ipynb) to [Cloud Run](https://cloud.google.com/run/docs/overview/what-is-cloud-run). Same FastAPI app, same Storage Read API — but running as a managed, autoscaling service with IAM authentication and revision-based traffic splitting.

**Story arc:**
1. Package the FastAPI app into a container and deploy to Cloud Run
2. Authenticate with ID tokens (not access tokens)
3. Benchmark: single-entity latency, batch scaling, combined-restriction optimization
4. Stress test: sustained throughput ramp and traffic spikes
5. Compare serving patterns: sequential, concurrent, per-entity batch, combined batch
6. Traffic splitting across revisions (canary deployment)
7. Production autoscaling configuration

**Container approach:** The app reads BigQuery configuration from environment variables (`BQ_PROJECT`, `BQ_DATASET`, `BQ_TABLE`, `FEATURE_COLUMNS`). No model files to download — it connects directly to BigQuery's storage layer at startup via the [Storage Read API](https://cloud.google.com/bigquery/docs/reference/storage).

**Prerequisites:** Run [BigQuery Feature Store — Environment](./BigQuery%20Feature%20Store%20-%20Environment.ipynb) first to create the dataset and tables. Familiarity with the [main BigQuery Feature Store notebook](./BigQuery%20Feature%20Store.ipynb) is recommended.

**Reference:** For the Cloud Run deployment pattern used here, see [Serving Models on Cloud Run](../../Serving/Platforms/Serving%20Models%20on%20Cloud%20Run.ipynb).

**What you'll learn:**

*Deployment & auth:*
- Containerizing a BigQuery Storage Read API app for Cloud Run
- Cloud Run authentication: ID tokens vs access tokens

*Serving patterns:*
- Single-entity reads with session caching
- Per-entity batch: N concurrent ReadSessions via `asyncio.gather()`
- Combined-restriction batch: 1 ReadSession for all entities via `IN (...)` clause
- When each pattern wins — entity count is the decision variable

*Performance at scale:*
- Cold start, warm request, and Cloud Run network overhead
- Concurrency scaling: throughput vs latency tradeoff
- Sustained load testing: throughput plateau and latency degradation under production-like traffic
- Spike bursts: autoscaling response and recovery latency

*Operations:*
- Traffic splitting between feature configurations (revision-based canary)
- Autoscaling tuning: `max_instance_request_concurrency`, instance count, CPU allocation

## 1. The Feature Serving App

The FastAPI app from the [main notebook](./BigQuery%20Feature%20Store.ipynb) (Part 5), packaged for Cloud Run. Key design:

- **`BigQueryReadClient()`** initialized once at startup — gRPC channel pool reused across all requests
- **`FeatureSessionPool`** for session caching (same class from the main notebook)
- **`asyncio.to_thread()`** for non-blocking reads in async endpoints
- **Per-entity batch** (`POST /features/batch`) — accepts multiple entities, reads concurrently via `asyncio.gather()`
- **Combined-restriction batch** (`POST /features/batch/combined`) — reads all entities in a single ReadSession using an `IN (...)` row restriction, eliminating N-1 session creation roundtrips
- **Configuration via environment variables** — `BQ_PROJECT`, `BQ_DATASET`, `BQ_TABLE`, `FEATURE_COLUMNS`

```python
%%writefile {DIR}/source/app/main.py
import os
import re
import time
import asyncio
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from google.cloud import bigquery_storage
from google.cloud.bigquery_storage import types as bqs_types
import pyarrow as pa

# --- Configuration from environment ---
BQ_PROJECT = os.environ['BQ_PROJECT']
BQ_DATASET = os.environ['BQ_DATASET']
BQ_TABLE = os.environ['BQ_TABLE']
FEATURE_COLUMNS = os.environ.get(
    'FEATURE_COLUMNS',
    'feature_float_1,feature_float_2,feature_float_3,feature_float_4,feature_float_5'
).split(',')
TABLE_PATH = f'projects/{BQ_PROJECT}/datasets/{BQ_DATASET}/tables/{BQ_TABLE}'


class FeatureSessionPool:
    def __init__(self, client, table, project, selected_fields, ttl_seconds=300):
        self.client = client
        self.table = table
        self.project = project
        self.selected_fields = selected_fields
        self.ttl_seconds = ttl_seconds
        self._cache = {}

    def get_session(self, row_restriction):
        self._evict_expired()
        if row_restriction in self._cache:
            return self._cache[row_restriction][0]
        session = self._create(row_restriction)
        self._cache[row_restriction] = (session, time.time())
        return session

    def _create(self, row_restriction):
        return self.client.create_read_session(
            request=dict(
                parent=f'projects/{self.project}',
                read_session=dict(
                    table=self.table,
                    data_format=bqs_types.DataFormat.ARROW,
                    read_options=dict(
                        selected_fields=self.selected_fields,
                        row_restriction=row_restriction,
                    ),
                ),
                max_stream_count=1,
            )
        )

    def _evict_expired(self):
        now = time.time()
        expired = [k for k, (_, t) in self._cache.items() if now - t > self.ttl_seconds]
        for k in expired:
            del self._cache[k]

    @property
    def cache_size(self):
        return len(self._cache)


# --- App ---
app = FastAPI(title='BigQuery Feature Store')
client = bigquery_storage.BigQueryReadClient()
pool = FeatureSessionPool(
    client=client,
    table=TABLE_PATH,
    project=BQ_PROJECT,
    selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS,
    ttl_seconds=300,
)

ENTITY_PATTERN = re.compile(r'^[A-Z]$')
ID_PATTERN = re.compile(r'^\d{6}$')


def read_features_sync(entity_group: str, entity_id: str) -> dict:
    restriction = f"entity_group = '{entity_group}' AND entity_id = '{entity_id}'"
    session = pool.get_session(restriction)
    if not session.streams:
        return {}
    reader = client.read_rows(session.streams[0].name)
    arrow_table = reader.to_arrow(session)
    if arrow_table.num_rows == 0:
        return {}
    return {col: arrow_table.column(col).to_pylist()[0] for col in arrow_table.column_names}


def read_features_batch_combined(entities: list[dict]) -> list[dict]:
    """Read multiple entities in a single ReadSession using a combined row_restriction.

    Builds: (entity_group, entity_id) IN (('A','000001'), ('B','000002'), ...)
    1 session + 1 stream instead of N — eliminates N-1 session creation roundtrips.
    """
    pairs = ', '.join(
        f"('{e['entity_group']}', '{e['entity_id']}')" for e in entities
    )
    restriction = f"(entity_group, entity_id) IN ({pairs})"

    session = client.create_read_session(
        request=dict(
            parent=f'projects/{BQ_PROJECT}',
            read_session=dict(
                table=TABLE_PATH,
                data_format=bqs_types.DataFormat.ARROW,
                read_options=dict(
                    selected_fields=['entity_group', 'entity_id'] + FEATURE_COLUMNS,
                    row_restriction=restriction,
                ),
            ),
            max_stream_count=1,
        )
    )
    if not session.streams:
        return []
    reader = client.read_rows(session.streams[0].name)
    arrow_table = reader.to_arrow(session)
    if arrow_table.num_rows == 0:
        return []
    return [
        {col: arrow_table.column(col).to_pylist()[i] for col in arrow_table.column_names}
        for i in range(arrow_table.num_rows)
    ]


@app.get('/features/{entity_group}/{entity_id}')
async def get_features(entity_group: str, entity_id: str):
    if not ENTITY_PATTERN.match(entity_group):
        raise HTTPException(400, 'entity_group must be a single uppercase letter')
    if not ID_PATTERN.match(entity_id):
        raise HTTPException(400, 'entity_id must be a 6-digit zero-padded number')
    features = await asyncio.to_thread(read_features_sync, entity_group, entity_id)
    if not features:
        raise HTTPException(404, 'Entity not found')
    return features


class BatchRequest(BaseModel):
    entities: list[dict]


@app.post('/features/batch')
async def get_features_batch(request: BatchRequest):
    """Per-entity batch: N individual reads via asyncio.gather."""
    for e in request.entities:
        if not ENTITY_PATTERN.match(e.get('entity_group', '')):
            raise HTTPException(400, f'Invalid entity_group: {e.get("entity_group")}')
        if not ID_PATTERN.match(e.get('entity_id', '')):
            raise HTTPException(400, f'Invalid entity_id: {e.get("entity_id")}')

    async def read_one(e):
        return await asyncio.to_thread(
            read_features_sync, e['entity_group'], e['entity_id']
        )

    results = await asyncio.gather(*[read_one(e) for e in request.entities])
    return {'features': [r for r in results if r]}


@app.post('/features/batch/combined')
async def get_features_batch_combined(request: BatchRequest):
    """Combined-restriction batch: 1 session for all entities."""
    for e in request.entities:
        if not ENTITY_PATTERN.match(e.get('entity_group', '')):
            raise HTTPException(400, f'Invalid entity_group: {e.get("entity_group")}')
        if not ID_PATTERN.match(e.get('entity_id', '')):
            raise HTTPException(400, f'Invalid entity_id: {e.get("entity_id")}')
    results = await asyncio.to_thread(read_features_batch_combined, request.entities)
    return {'features': results}


@app.get('/health')
async def health():
    return {
        'status': 'ok',
        'cached_sessions': pool.cache_size,
        'table': TABLE_PATH,
        'feature_columns': len(FEATURE_COLUMNS),
    }
```

## 4. Deploy to Cloud Run

Deploy the container as a Cloud Run service. Configuration via environment variables makes the container self-contained — it connects to BigQuery at startup using ADC (Application Default Credentials) provided by Cloud Run's service account.

| Aspect | Local (main notebook) | Cloud Run |
|--------|---|---|
| Startup | Instant (same process) | Container pull + init (~2-5s cold start) |
| Auth | ADC from local environment | ADC from Cloud Run service account |
| Scaling | Single process | Autoscaling (0 to N instances) |
| Network | Localhost | HTTPS + IAM (ID tokens) |
| Configuration | Python variables | Environment variables |

### The Right Way: ID Tokens

```python
def cloud_run_get(path, service_url=None):
    url = service_url or SERVICE_URL
    auth_req = google.auth.transport.requests.Request()
    id_token = google.oauth2.id_token.fetch_id_token(auth_req, url)
    resp = requests.get(
        f'{url}{path}',
        headers={'Authorization': f'Bearer {id_token}'},
        timeout=60,
    )
    resp.raise_for_status()
    return resp

def cloud_run_post(path, json_body, service_url=None):
    url = service_url or SERVICE_URL
    auth_req = google.auth.transport.requests.Request()
    id_token = google.oauth2.id_token.fetch_id_token(auth_req, url)
    resp = requests.post(
        f'{url}{path}',
        headers={
            'Authorization': f'Bearer {id_token}',
            'Content-Type': 'application/json',
        },
        json=json_body,
        timeout=60,
    )
    resp.raise_for_status()
    return resp

# Test with ID token
resp = cloud_run_get('/health')
print(f'Health check: {resp.status_code}')
print(f'Response: {resp.json()}')
```

## 7. Entity Count Scaling

How does latency scale with the number of entities per request? Two batch endpoints take different approaches:

- **Per-entity batch** (`POST /features/batch`) — N individual ReadSessions run concurrently server-side via `asyncio.gather()`. Latency scales sub-linearly but each entity still creates its own session.
- **Combined-restriction batch** (`POST /features/batch/combined`) — all entities in a single ReadSession using `(entity_group, entity_id) IN (...)`. One session regardless of entity count.

First, the per-entity batch — then the combined approach and how they compare.

```python
# Batch endpoint at varying entity counts
random.seed(42)
entity_counts = [1, 5, 10, 25, 50, 100]
batch_results = []

for n in entity_counts:
    entities = [
        {'entity_group': chr(ord('A') + i % 26), 'entity_id': str(i + 1).zfill(6)}
        for i in range(n)
    ]

    start = time.perf_counter()
    resp = cloud_run_post('/features/batch', {'entities': entities})
    latency_ms = (time.perf_counter() - start) * 1000
    result = resp.json()

    batch_results.append((n, latency_ms, len(result['features'])))
    print(f'  Batch {n:>3d} entities: {latency_ms:>7.0f} ms  ({len(result["features"])} returned)')

print(f'\nServer-side async reads keep latency sub-linear with entity count.')
```

```python
# Sequential vs batch comparison
N_COMPARE = 10
entities = [
    {'entity_group': chr(ord('A') + i % 26), 'entity_id': str(i + 1).zfill(6)}
    for i in range(N_COMPARE)
]

# Sequential: N individual GET requests
start = time.perf_counter()
for e in entities:
    cloud_run_get(f'/features/{e["entity_group"]}/{e["entity_id"]}')
sequential_ms = (time.perf_counter() - start) * 1000

# Batch: 1 POST request with N entities
start = time.perf_counter()
cloud_run_post('/features/batch', {'entities': entities})
batch_ms = (time.perf_counter() - start) * 1000

print(f'{N_COMPARE} entity reads — sequential vs batch:')
print(f'  Sequential ({N_COMPARE}x GET):  {sequential_ms:>7.0f} ms  ({sequential_ms/N_COMPARE:.0f} ms/entity)')
print(f'  Batch (1x POST):        {batch_ms:>7.0f} ms  ({batch_ms/N_COMPARE:.0f} ms/entity)')
print(f'  Speedup:                {sequential_ms / batch_ms:.1f}x')
print(f'\nBatching eliminates per-request network roundtrips through Cloud Run.')
```

```python
# Visualize entity count scaling
fig, ax = plt.subplots(figsize=(10, 5))

counts = [r[0] for r in batch_results]
latencies = [r[1] for r in batch_results]

bars = ax.bar(range(len(counts)), latencies, color='#1565C0')
for bar, val in zip(bars, latencies):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(latencies) * 0.02,
            f'{val:.0f}ms', ha='center', va='bottom', fontsize=9, fontweight='bold')

# Add sequential reference line for comparison
per_entity_sequential = sequential_ms / N_COMPARE
seq_reference = [per_entity_sequential * n for n in counts]
ax.plot(range(len(counts)), seq_reference, 'r--', linewidth=2, label=f'Sequential equivalent ({per_entity_sequential:.0f}ms/entity)')

ax.set_xlabel('Entities per Batch Request', fontweight='bold')
ax.set_ylabel('Latency (ms)', fontweight='bold')
ax.set_title('Cloud Run: Batch Endpoint Latency vs Entity Count', fontsize=13, fontweight='bold')
ax.set_xticks(range(len(counts)))
ax.set_xticklabels([str(c) for c in counts])
ax.legend()
plt.tight_layout()
plt.show()

print(f'Batch endpoint scales sub-linearly — server-side async reads amortize BigQuery session overhead.')
```

### Combined-Restriction Batch Optimization

The per-entity batch endpoint (`POST /features/batch`) creates **N separate ReadSessions** — one per entity. Each `create_read_session()` call takes ~200ms. At 100 entities, that's ~20 seconds of session creation alone.

The combined batch endpoint (`POST /features/batch/combined`) builds a single `(entity_group, entity_id) IN (...)` row restriction and creates **1 ReadSession** for all entities. Same results, dramatically fewer API calls.

```python
# Per-entity batch vs combined batch at varying entity counts
compare_counts = [1, 5, 10, 25, 50, 100]
per_entity_latencies = []
combined_latencies = []

for n in compare_counts:
    entities = [
        {'entity_group': chr(ord('A') + i % 26), 'entity_id': str(i + 1).zfill(6)}
        for i in range(n)
    ]

    # Per-entity batch: N sessions via asyncio.gather
    start = time.perf_counter()
    resp = cloud_run_post('/features/batch', {'entities': entities})
    pe_ms = (time.perf_counter() - start) * 1000
    per_entity_latencies.append(pe_ms)

    # Combined batch: 1 session via IN restriction
    start = time.perf_counter()
    resp = cloud_run_post('/features/batch/combined', {'entities': entities})
    cb_ms = (time.perf_counter() - start) * 1000
    combined_latencies.append(cb_ms)

    speedup = pe_ms / cb_ms if cb_ms > 0 else float('inf')
    print(f'  {n:>3d} entities:  per-entity={pe_ms:>7.0f}ms  combined={cb_ms:>7.0f}ms  speedup={speedup:.1f}x')

print(f'\nCombined batch uses 1 ReadSession instead of N — the speedup grows with entity count.')
```

```python
# Visualize per-entity vs combined batch
fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(len(compare_counts))
width = 0.35

bars1 = ax.bar(x - width/2, per_entity_latencies, width, color='#1565C0', label='Per-entity batch (N sessions)')
bars2 = ax.bar(x + width/2, combined_latencies, width, color='#FF8F00', label='Combined batch (1 session)')

for bar, val in zip(bars1, per_entity_latencies):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(per_entity_latencies) * 0.02,
            f'{val:.0f}ms', ha='center', va='bottom', fontsize=8, fontweight='bold', color='#1565C0')
for bar, val in zip(bars2, combined_latencies):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(per_entity_latencies) * 0.02,
            f'{val:.0f}ms', ha='center', va='bottom', fontsize=8, fontweight='bold', color='#FF8F00')

ax.set_xlabel('Entities per Batch Request', fontweight='bold')
ax.set_ylabel('Latency (ms)', fontweight='bold')
ax.set_title('Cloud Run: Per-Entity Batch vs Combined Batch', fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels([str(c) for c in compare_counts])
ax.legend()
plt.tight_layout()
plt.show()

print(f'Combined batch eliminates N-1 session creation roundtrips (~200ms each).')
print(f'The advantage grows with entity count — at 100 entities, it\'s dramatic.')
```

## 8. Concurrent Request Scaling

How does throughput scale as we increase concurrent in-flight requests? This tests Cloud Run's autoscaling — as load increases, Cloud Run spins up more container instances (up to `max_instance_count`).

```python
# Concurrency scaling: vary concurrent workers
concurrency_levels = [1, 5, 10, 20, 50]
N_REQUESTS = 50
concurrency_results = []

for workers in concurrency_levels:
    def timed_request(i):
        group = chr(ord('A') + i % 26)
        entity = str(i + 1).zfill(6)
        start = time.perf_counter()
        cloud_run_get(f'/features/{group}/{entity}')
        return (time.perf_counter() - start) * 1000

    wall_start = time.perf_counter()
    with ThreadPoolExecutor(max_workers=workers) as executor:
        results = list(executor.map(timed_request, range(N_REQUESTS)))
    wall_ms = (time.perf_counter() - wall_start) * 1000

    latencies = sorted(results)
    throughput = N_REQUESTS / (wall_ms / 1000)
    p50 = latencies[len(latencies) // 2]
    p95 = latencies[int(len(latencies) * 0.95)]

    concurrency_results.append((workers, throughput, p50, p95, wall_ms))
    print(f'  Workers={workers:>2d}:  throughput={throughput:>5.1f} req/s  '
          f'p50={p50:>6.0f}ms  p95={p95:>6.0f}ms  wall={wall_ms:>7.0f}ms')

print(f'\n{N_REQUESTS} total requests at each concurrency level.')
```

```python
# Visualize concurrency scaling
fig, ax1 = plt.subplots(figsize=(10, 5))

workers_list = [r[0] for r in concurrency_results]
throughputs = [r[1] for r in concurrency_results]
p50s = [r[2] for r in concurrency_results]
p95s = [r[3] for r in concurrency_results]

color_throughput = '#1565C0'
color_latency = '#E53935'

ax1.set_xlabel('Concurrent Workers', fontweight='bold')
ax1.set_ylabel('Throughput (req/s)', fontweight='bold', color=color_throughput)
ax1.plot(workers_list, throughputs, 'o-', color=color_throughput, linewidth=2, markersize=8, label='Throughput')
ax1.tick_params(axis='y', labelcolor=color_throughput)

ax2 = ax1.twinx()
ax2.set_ylabel('Latency (ms)', fontweight='bold', color=color_latency)
ax2.plot(workers_list, p50s, 's--', color=color_latency, linewidth=1.5, markersize=6, label='p50 latency')
ax2.plot(workers_list, p95s, '^--', color='#FF7043', linewidth=1.5, markersize=6, label='p95 latency')
ax2.tick_params(axis='y', labelcolor=color_latency)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

ax1.set_title('Cloud Run: Concurrency Scaling — Throughput vs Latency', fontsize=13, fontweight='bold')
ax1.set_xticks(workers_list)
plt.tight_layout()
plt.show()

print(f'Cloud Run autoscaling adds instances as concurrency increases.')
print(f'Throughput scales until the BigQuery Storage API or Cloud Run instance limit saturates.')
```

```python
# Concurrent batch requests: multiple callers each requesting multiple entities
batch_concurrency_levels = [1, 5, 10, 20]
ENTITIES_PER_BATCH = 10
N_BATCH_REQUESTS = 20
batch_concurrency_results = []

for workers in batch_concurrency_levels:
    def timed_batch_request(i):
        entities = [
            {'entity_group': chr(ord('A') + (i * ENTITIES_PER_BATCH + j) % 26),
             'entity_id': str((i * ENTITIES_PER_BATCH + j) % 5000 + 1).zfill(6)}
            for j in range(ENTITIES_PER_BATCH)
        ]
        start = time.perf_counter()
        cloud_run_post('/features/batch', {'entities': entities})
        return (time.perf_counter() - start) * 1000

    wall_start = time.perf_counter()
    with ThreadPoolExecutor(max_workers=workers) as executor:
        results = list(executor.map(timed_batch_request, range(N_BATCH_REQUESTS)))
    wall_ms = (time.perf_counter() - wall_start) * 1000

    latencies = sorted(results)
    total_entities = N_BATCH_REQUESTS * ENTITIES_PER_BATCH
    entity_throughput = total_entities / (wall_ms / 1000)
    p50 = latencies[len(latencies) // 2]
    p95 = latencies[int(len(latencies) * 0.95)]

    batch_concurrency_results.append((workers, entity_throughput, p50, p95))
    print(f'  Workers={workers:>2d}:  {entity_throughput:>6.0f} entities/s  '
          f'p50={p50:>6.0f}ms  p95={p95:>6.0f}ms  ({N_BATCH_REQUESTS}x{ENTITIES_PER_BATCH} entities)')

print(f'\nConcurrent batch requests: the realistic production pattern.')
```

### Sustained Load & Spike Testing

Section 8 tested 50 requests at varying concurrency — enough to see the throughput curve but not enough to stress the system. Production feature stores handle sustained high-volume traffic with sudden spikes (batch inference jobs, flash sales, model retraining pipelines requesting features).

Here we push harder:
1. **Sustained throughput ramp** — 50 to 1,000 total requests at 50 concurrent workers. Where does throughput plateau? When do latency tails blow up?
2. **Spike burst** — 10 to 200 simultaneous requests hitting the service at once. How does Cloud Run autoscaling respond? How fast does latency recover?

Current deploy: `max_instance_count=5`, default `concurrency=80` — up to 400 concurrent requests on paper.

```python
# Sustained throughput ramp: increase total requests at fixed high concurrency
WORKERS = 50
request_counts = [50, 100, 200, 500, 1000]
sustained_results = []

def timed_request_safe(i):
    """Fire a single GET request, return (latency_ms, success)."""
    group = chr(ord('A') + i % 26)
    entity = str(i % 5000 + 1).zfill(6)
    start = time.perf_counter()
    try:
        resp = cloud_run_get(f'/features/{group}/{entity}')
        return ((time.perf_counter() - start) * 1000, True)
    except Exception:
        return ((time.perf_counter() - start) * 1000, False)

print(f'Sustained throughput ramp ({WORKERS} concurrent workers)\n')
print(f'  {"Requests":>10s}  {"Throughput":>12s}  {"p50":>8s}  {"p95":>8s}  {"p99":>8s}  {"Errors":>8s}  {"Success":>8s}')
print(f'  {"-"*10}  {"-"*12}  {"-"*8}  {"-"*8}  {"-"*8}  {"-"*8}  {"-"*8}')

for n_requests in request_counts:
    wall_start = time.perf_counter()
    with ThreadPoolExecutor(max_workers=WORKERS) as executor:
        results = list(executor.map(timed_request_safe, range(n_requests)))
    wall_ms = (time.perf_counter() - wall_start) * 1000

    latencies = sorted([r[0] for r in results if r[1]])
    errors = sum(1 for r in results if not r[1])
    throughput = n_requests / (wall_ms / 1000)

    if latencies:
        p50 = latencies[len(latencies) // 2]
        p95 = latencies[int(len(latencies) * 0.95)]
        p99 = latencies[int(len(latencies) * 0.99)]
    else:
        p50 = p95 = p99 = float('nan')

    success_pct = (n_requests - errors) / n_requests * 100
    sustained_results.append((n_requests, throughput, p50, p95, p99, errors, success_pct))

    print(f'  {n_requests:>10d}  {throughput:>10.1f}/s  {p50:>6.0f}ms  {p95:>6.0f}ms  {p99:>6.0f}ms  {errors:>8d}  {success_pct:>6.1f}%')

print(f'\nThroughput plateau shows where Cloud Run instances + BigQuery Storage API saturate.')
print(f'Latency tails (p95, p99) reveal queueing under load.')
```

```python
# Spike burst test: N simultaneous requests, then measure recovery
spike_sizes = [10, 25, 50, 100, 200]
spike_results = []

print(f'Spike burst test\n')
print(f'  {"Spike":>8s}  {"p50":>8s}  {"p95":>8s}  {"Max":>8s}  {"Errors":>8s}  {"Recovery":>10s}')
print(f'  {"-"*8}  {"-"*8}  {"-"*8}  {"-"*8}  {"-"*8}  {"-"*10}')

for spike_size in spike_sizes:
    # Warm-up: ensure at least 1 instance is active
    try:
        cloud_run_get('/features/A/000001')
    except Exception:
        pass
    time.sleep(0.5)

    # Spike: all requests launch simultaneously
    with ThreadPoolExecutor(max_workers=spike_size) as executor:
        results = list(executor.map(timed_request_safe, range(spike_size)))

    spike_latencies = sorted([r[0] for r in results if r[1]])
    spike_errors = sum(1 for r in results if not r[1])

    if spike_latencies:
        sp50 = spike_latencies[len(spike_latencies) // 2]
        sp95 = spike_latencies[int(len(spike_latencies) * 0.95)]
        sp_max = spike_latencies[-1]
    else:
        sp50 = sp95 = sp_max = float('nan')

    # Recovery: 5 sequential requests after the burst
    recovery_latencies = []
    for i in range(5):
        start = time.perf_counter()
        try:
            cloud_run_get(f'/features/{chr(ord("A") + i % 26)}/{str(i + 1).zfill(6)}')
            recovery_latencies.append((time.perf_counter() - start) * 1000)
        except Exception:
            pass

    recovery_p50 = sorted(recovery_latencies)[len(recovery_latencies) // 2] if recovery_latencies else float('nan')

    spike_results.append((spike_size, sp50, sp95, sp_max, spike_errors, recovery_p50))
    print(f'  {spike_size:>8d}  {sp50:>6.0f}ms  {sp95:>6.0f}ms  {sp_max:>6.0f}ms  {spike_errors:>8d}  {recovery_p50:>8.0f}ms')

print(f'\nSmall spikes are absorbed by warm instances.')
print(f'Large spikes trigger autoscaling — new instances add cold start latency.')
print(f'Recovery latency shows how fast the service returns to baseline.')
```

## 9. Sync vs Async Comparison

Four patterns for reading 10 entities through Cloud Run, from most to fewest network roundtrips:

1. **Sync sequential** — 10 individual `GET` requests, one after another
2. **Sync concurrent** — 10 individual `GET` requests via `ThreadPoolExecutor`
3. **Per-entity batch** — 1 `POST /features/batch` request (server-side `asyncio.gather`, N sessions)
4. **Combined batch** — 1 `POST /features/batch/combined` request (single ReadSession with `IN (...)` restriction)

```python
N_ENTITIES = 10
entities = [
    {'entity_group': chr(ord('A') + i % 26), 'entity_id': str(i + 1).zfill(6)}
    for i in range(N_ENTITIES)
]

# Pattern 1: Sync sequential
start = time.perf_counter()
for e in entities:
    cloud_run_get(f'/features/{e["entity_group"]}/{e["entity_id"]}')
sync_sequential_ms = (time.perf_counter() - start) * 1000

# Pattern 2: Sync concurrent
def fetch_one(e):
    return cloud_run_get(f'/features/{e["entity_group"]}/{e["entity_id"]}')

start = time.perf_counter()
with ThreadPoolExecutor(max_workers=N_ENTITIES) as executor:
    list(executor.map(fetch_one, entities))
sync_concurrent_ms = (time.perf_counter() - start) * 1000

# Pattern 3: Per-entity batch (N sessions via asyncio.gather)
start = time.perf_counter()
cloud_run_post('/features/batch', {'entities': entities})
per_entity_batch_ms = (time.perf_counter() - start) * 1000

# Pattern 4: Combined batch (1 session via IN restriction)
start = time.perf_counter()
cloud_run_post('/features/batch/combined', {'entities': entities})
combined_batch_ms = (time.perf_counter() - start) * 1000

print(f'{N_ENTITIES} entity reads — four serving patterns:')
print(f'  {"Pattern":<30s} {"Total":>10s} {"Per Entity":>12s} {"Speedup":>10s}')
print(f'  {"-"*30} {"-"*10} {"-"*12} {"-"*10}')
print(f'  {"Sync sequential":<30s} {sync_sequential_ms:>9.0f}ms {sync_sequential_ms/N_ENTITIES:>10.0f}ms {"1.0x":>10s}')
print(f'  {"Sync concurrent":<30s} {sync_concurrent_ms:>9.0f}ms {sync_concurrent_ms/N_ENTITIES:>10.0f}ms {sync_sequential_ms/sync_concurrent_ms:>9.1f}x')
print(f'  {"Per-entity batch (POST)":<30s} {per_entity_batch_ms:>9.0f}ms {per_entity_batch_ms/N_ENTITIES:>10.0f}ms {sync_sequential_ms/per_entity_batch_ms:>9.1f}x')
print(f'  {"Combined batch (POST)":<30s} {combined_batch_ms:>9.0f}ms {combined_batch_ms/N_ENTITIES:>10.0f}ms {sync_sequential_ms/combined_batch_ms:>9.1f}x')

print(f'\nAt 10 entities, per-entity and combined batch are close — combined\'s advantage grows with count.')
print(f'Per-entity batch: 1 roundtrip, N server-side sessions. Combined: 1 roundtrip, 1 session.')
print(f'For bulk reads (25+ entities), combined batch dominates. For small batches, either works.')
```

## 12. Autoscaling Configuration

Cloud Run scales based on **concurrent requests per instance** (default: 80). Key settings for feature serving:

| Setting | Default | Recommendation |
|---------|---------|---------------|
| `min_instance_count` | 0 (scale-to-zero) | 1 for production (avoids cold starts) |
| `max_instance_count` | 100 | Set based on expected peak QPS |
| `max_instance_request_concurrency` | 80 | Lower (10-20) if BigQuery gRPC channels saturate |
| CPU allocation | Per-request | Always-on for consistent latency |

The BigQuery Storage Read API client uses a gRPC channel pool internally. At high concurrency, the channel pool can become a bottleneck. Setting `max_instance_request_concurrency` lower forces Cloud Run to add more instances earlier, each with its own client and channel pool.

The sustained load tests in Section 8 show throughput plateauing around 30-40 req/s with the default configuration (`max_instance_count=5`, `concurrency=80`), and p95 latency crossing 1 second under sustained load. Spike bursts up to 200 simultaneous requests were absorbed with zero errors, but latency degraded significantly. These results motivate the production recommendations below.

```python
# Example: production-ready autoscaling configuration
service = cr.get_service(name=service_path)

service.template.scaling = run_v2.RevisionScaling(
    min_instance_count=1,    # keep 1 warm (avoid cold starts)
    max_instance_count=10,   # scale up to 10 instances
)
service.template.max_instance_request_concurrency = 20  # lower than default 80

# Note: uncommenting the update below will create a new revision
# operation = cr.update_service(service=service)
# service_resp = operation.result()

print('Production autoscaling configuration (not applied — uncomment to apply):')
print(f'  min_instance_count: 1')
print(f'  max_instance_count: 10')
print(f'  max_instance_request_concurrency: 20')
print(f'\nWith 10 instances x 20 concurrent = 200 concurrent requests capacity.')
print(f'Each instance has its own BigQueryReadClient + gRPC channel pool.')
```
