# BigQuery Feature Store — Environment

This notebook creates the BigQuery dataset and synthetic feature tables used by all notebooks in this series.

**Resources created:**
- BigQuery dataset: `bigquery_feature_store`
- Table: `dense_features` — 130,000 rows, 224 columns (all BigQuery data types)
- Table: `sparse_events` — 2,600,000 rows (time-series events per entity)
- Table: `dense_features_clustered` — clustered copy for Storage Read API performance
- Table: `entity_embeddings` — 64-dim vectors for similarity search demos

**Compound entity key:** All data uses a two-part key (`entity_group` + `entity_id`) — 26 groups (A–Z) with 5,000 entities per group (130,000 total).

## 2. Dense Features Table

The dense table provides **one row per entity** with 200+ feature columns covering every [BigQuery data type](https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types). This is the primary table used for feature serving demonstrations.

### Design

| Aspect | Value |
|--------|-------|
| **Entity groups** | 26 groups ("A" through "Z") |
| **Entities per group** | 5,000 (IDs: "000001" through "005000") |
| **Total rows** | 130,000 |
| **Compound key** | `entity_group` + `entity_id` |
| **Label** | Binary `label` column derived from feature values — has learnable signal for model training |

### Data Type Coverage

| BQ Data Type | Column Prefix | Count |
|-------------|---------------|-------|
| INT64 | `feature_int_` | 5 |
| FLOAT64 | `feature_float_` | 180 |
| NUMERIC | `feature_numeric_` | 3 |
| BIGNUMERIC | `feature_bignumeric_` | 2 |
| STRING | `feature_string_` | 5 |
| BYTES | `feature_bytes_` | 2 |
| BOOL | `feature_bool_` | 3 |
| DATE | `feature_date_` | 3 |
| DATETIME | `feature_datetime_` | 2 |
| TIME | `feature_time_` | 2 |
| TIMESTAMP | `feature_timestamp_` | 3 |
| GEOGRAPHY | `feature_geo_` | 2 |
| JSON | `feature_json_` | 2 |
| ARRAY<INT64> | `feature_array_int_` | 2 |
| ARRAY<STRING> | `feature_array_string_` | 1 |
| STRUCT | `feature_struct_` | 2 |
| ARRAY<STRUCT> | `feature_array_struct_` | 1 |
| RANGE<DATE> | `feature_range_date_` | 1 |

Five columns have ~5% NULL values: `feature_float_3`, `feature_float_4`, `feature_string_2`, `feature_date_2`, `feature_numeric_2`.

````python
DENSE_TABLE_ID = f'{BQ_PROJECT}.{BQ_DATASET}.{BQ_TABLE_DENSE}'

print(f'Creating {DENSE_TABLE_ID}...')
print(f'This may take 30-60 seconds.\n')

# Float columns (feature_float_6 through feature_float_180)
float_columns = []
for i in range(6, 181):
    if i % 5 == 0:
        float_columns.append(f'    ROUND(RAND() * 1000, 4) AS feature_float_{i}')
    elif i % 5 == 1:
        float_columns.append(f'    ROUND(RAND() * 100 - 50, 4) AS feature_float_{i}')
    elif i % 5 == 2:
        float_columns.append(f'    ROUND(RAND() * 10, 4) AS feature_float_{i}')
    elif i % 5 == 3:
        float_columns.append(f'    ROUND(RAND() * 500 + 100, 4) AS feature_float_{i}')
    else:
        float_columns.append(f'    ROUND(RAND(), 6) AS feature_float_{i}')

float_cols_sql = ',\n'.join(float_columns)

dense_query = f"""
CREATE OR REPLACE TABLE `{DENSE_TABLE_ID}` AS
WITH
entity_groups AS (
    SELECT CHR(code) AS entity_group
    FROM UNNEST(GENERATE_ARRAY(65, 90)) AS code
),
entity_ids AS (
    SELECT FORMAT('%06d', id) AS entity_id
    FROM UNNEST(GENERATE_ARRAY(1, 5000)) AS id
),
entities AS (
    SELECT entity_group, entity_id
    FROM entity_groups CROSS JOIN entity_ids
),
raw_features AS (
    SELECT
        entity_group,
        entity_id,

        -- INT64 features (5)
        CAST(FLOOR(RAND() * 1000) AS INT64) AS feature_int_1,
        CAST(FLOOR(RAND() * 10000) AS INT64) AS feature_int_2,
        CAST(FLOOR(RAND() * 100) AS INT64) AS feature_int_3,
        CAST(FLOOR(RAND() * 1000000) AS INT64) AS feature_int_4,
        CAST(MOD(ABS(FARM_FINGERPRINT(CONCAT(entity_group, entity_id))), 500) AS INT64) AS feature_int_5,

        -- FLOAT64 features (first 5)
        ROUND(RAND() * 100, 4) AS feature_float_1,
        ROUND(RAND() * 1000, 4) AS feature_float_2,
        IF(RAND() > 0.05, ROUND(RAND() * 50 - 25, 4), NULL) AS feature_float_3,
        IF(RAND() > 0.05, ROUND(RAND() * 10, 4), NULL) AS feature_float_4,
        ROUND(RAND(), 6) AS feature_float_5,

        -- NUMERIC features (3)
        CAST(ROUND(RAND() * 99999, 2) AS NUMERIC) AS feature_numeric_1,
        IF(RAND() > 0.05, CAST(ROUND(RAND() * 9999999, 4) AS NUMERIC), NULL) AS feature_numeric_2,
        CAST(ROUND(RAND() * 100, 6) AS NUMERIC) AS feature_numeric_3,

        -- BIGNUMERIC features (2)
        CAST(ROUND(RAND() * 999999999999, 2) AS BIGNUMERIC) AS feature_bignumeric_1,
        CAST(ROUND(RAND() * 99999999, 8) AS BIGNUMERIC) AS feature_bignumeric_2,

        -- STRING features (5)
        CONCAT('cat_', CAST(CAST(FLOOR(RAND() * 20) AS INT64) AS STRING)) AS feature_string_1,
        IF(RAND() > 0.05, CONCAT('type_', CAST(CAST(FLOOR(RAND() * 50) AS INT64) AS STRING)), NULL) AS feature_string_2,
        ['alpha', 'beta', 'gamma', 'delta', 'epsilon', 'zeta', 'eta', 'theta'][ORDINAL(CAST(CEIL(RAND() * 8) AS INT64))] AS feature_string_3,
        CONCAT(entity_group, '_', ['north', 'south', 'east', 'west'][ORDINAL(CAST(CEIL(RAND() * 4) AS INT64))]) AS feature_string_4,
        FORMAT('%s-%06d', entity_group, CAST(FLOOR(RAND() * 999999) AS INT64)) AS feature_string_5,

        -- BYTES features (2)
        CAST(CONCAT('bytes_', entity_group, entity_id) AS BYTES) AS feature_bytes_1,
        CAST(FORMAT('%06d', CAST(FLOOR(RAND() * 999999) AS INT64)) AS BYTES) AS feature_bytes_2,

        -- BOOL features (3)
        RAND() > 0.5 AS feature_bool_1,
        RAND() > 0.7 AS feature_bool_2,
        MOD(CAST(FLOOR(RAND() * 10) AS INT64), 2) = 0 AS feature_bool_3,

        -- DATE features (3)
        DATE_ADD(DATE '2020-01-01', INTERVAL CAST(FLOOR(RAND() * 1825) AS INT64) DAY) AS feature_date_1,
        IF(RAND() > 0.05, DATE_ADD(DATE '2022-06-01', INTERVAL CAST(FLOOR(RAND() * 730) AS INT64) DAY), NULL) AS feature_date_2,
        DATE_ADD(DATE '2023-01-01', INTERVAL CAST(FLOOR(RAND() * 365) AS INT64) DAY) AS feature_date_3,

        -- DATETIME features (2)
        DATETIME_ADD(DATETIME '2023-01-01 00:00:00', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_datetime_1,
        DATETIME_ADD(DATETIME '2024-01-01 00:00:00', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_datetime_2,

        -- TIME features (2)
        TIME_ADD(TIME '00:00:00', INTERVAL CAST(FLOOR(RAND() * 86400) AS INT64) SECOND) AS feature_time_1,
        TIME_ADD(TIME '00:00:00', INTERVAL CAST(FLOOR(RAND() * 86400) AS INT64) SECOND) AS feature_time_2,

        -- TIMESTAMP features (3)
        TIMESTAMP_ADD(TIMESTAMP '2023-01-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_timestamp_1,
        TIMESTAMP_ADD(TIMESTAMP '2024-01-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_timestamp_2,
        TIMESTAMP_ADD(TIMESTAMP '2024-06-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 262800) AS INT64) MINUTE) AS feature_timestamp_3,

        -- GEOGRAPHY features (2)
        ST_GEOGPOINT(
            ROUND(-180 + RAND() * 360, 6),
            ROUND(-90 + RAND() * 180, 6)
        ) AS feature_geo_1,
        ST_GEOGPOINT(
            ROUND(-122.5 + RAND() * 5, 6),
            ROUND(37.0 + RAND() * 3, 6)
        ) AS feature_geo_2,

        -- JSON features (2)
        JSON_OBJECT(
            'score', ROUND(RAND() * 100, 2),
            'label', CONCAT('class_', CAST(CAST(FLOOR(RAND() * 10) AS INT64) AS STRING)),
            'active', RAND() > 0.5
        ) AS feature_json_1,
        JSON_OBJECT(
            'x', ROUND(RAND() * 10, 4),
            'y', ROUND(RAND() * 10, 4),
            'tag', entity_group
        ) AS feature_json_2,

        -- ARRAY<INT64> features (2)
        [CAST(FLOOR(RAND() * 100) AS INT64), CAST(FLOOR(RAND() * 100) AS INT64), CAST(FLOOR(RAND() * 100) AS INT64)] AS feature_array_int_1,
        [CAST(FLOOR(RAND() * 1000) AS INT64), CAST(FLOOR(RAND() * 1000) AS INT64)] AS feature_array_int_2,

        -- ARRAY<STRING> features (1)
        [
            ['red', 'blue', 'green', 'yellow'][ORDINAL(CAST(CEIL(RAND() * 4) AS INT64))],
            ['circle', 'square', 'triangle'][ORDINAL(CAST(CEIL(RAND() * 3) AS INT64))]
        ] AS feature_array_string_1,

        -- STRUCT features (2)
        STRUCT(
            ROUND(RAND() * 100, 2) AS value,
            CONCAT('unit_', CAST(CAST(FLOOR(RAND() * 5) AS INT64) AS STRING)) AS unit,
            RAND() > 0.5 AS is_valid
        ) AS feature_struct_1,
        STRUCT(
            CAST(FLOOR(RAND() * 1000) AS INT64) AS count,
            ROUND(RAND(), 4) AS ratio
        ) AS feature_struct_2,

        -- ARRAY<STRUCT> features (1)
        [
            STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name),
            STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name),
            STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name)
        ] AS feature_array_struct_1,

        -- RANGE<DATE> features (1)
        RANGE(
            DATE_ADD(DATE '2020-01-01', INTERVAL CAST(FLOOR(RAND() * 1000) AS INT64) DAY),
            DATE_ADD(DATE '2023-01-01', INTERVAL CAST(FLOOR(RAND() * 730) AS INT64) DAY)
        ) AS feature_range_date_1,

        -- Bulk FLOAT64 features (feature_float_6 through feature_float_180)
    {float_cols_sql}

    FROM entities
)
SELECT *,
    CASE WHEN (
        0.4 * feature_int_1 / 999.0 +
        0.3 * feature_float_1 / 100.0 +
        0.15 * (COALESCE(feature_float_3, 0.0) + 25.0) / 50.0 +
        0.15 * CAST(feature_bool_1 AS INT64)
        + (RAND() - 0.5) * 0.3
    ) > 0.6 THEN 'positive' ELSE 'negative' END AS label
FROM raw_features
"""

start = time.perf_counter()
bq.query(dense_query).result()
elapsed = time.perf_counter() - start
print(f'Dense table created in {elapsed:.1f}s')

info = bq.get_table(DENSE_TABLE_ID)
print(f'  Rows:    {info.num_rows:,}')
print(f'  Columns: {len(info.schema)}')
print(f'  Size:    {info.num_bytes / 1024**2:.1f} MB')
````

````python
print('=== Dense Table Verification ===\n')

# Row count and column count
info = bq.get_table(DENSE_TABLE_ID)
print(f'Rows:    {info.num_rows:,}')
print(f'Columns: {len(info.schema)}')

# Entity group distribution
eg_dist = bq.query(f'''
    SELECT entity_group, COUNT(*) AS n
    FROM `{DENSE_TABLE_ID}`
    GROUP BY entity_group
    ORDER BY entity_group
    LIMIT 5
''').to_dataframe()
print(f'\nEntity groups (first 5):')
print(eg_dist.to_string(index=False))

# Label distribution
label_dist = bq.query(f'''
    SELECT label, COUNT(*) AS n,
           ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(), 1) AS pct
    FROM `{DENSE_TABLE_ID}`
    GROUP BY label ORDER BY label
''').to_dataframe()
print(f'\nLabel distribution:')
print(label_dist.to_string(index=False))

# NULL rates
null_rates = bq.query(f'''
    SELECT
        ROUND(COUNTIF(feature_float_3 IS NULL) * 100.0 / COUNT(*), 1) AS float_3_null_pct,
        ROUND(COUNTIF(feature_float_4 IS NULL) * 100.0 / COUNT(*), 1) AS float_4_null_pct,
        ROUND(COUNTIF(feature_string_2 IS NULL) * 100.0 / COUNT(*), 1) AS string_2_null_pct,
        ROUND(COUNTIF(feature_date_2 IS NULL) * 100.0 / COUNT(*), 1) AS date_2_null_pct,
        ROUND(COUNTIF(feature_numeric_2 IS NULL) * 100.0 / COUNT(*), 1) AS numeric_2_null_pct
    FROM `{DENSE_TABLE_ID}`
''').to_dataframe()
print(f'\nNULL rates (~5% expected):')
print(null_rates.T.to_string(header=False))

# Data type inventory
type_inventory = bq.query(f'''
    SELECT data_type, COUNT(*) AS column_count
    FROM `{BQ_PROJECT}.{BQ_DATASET}.INFORMATION_SCHEMA.COLUMNS`
    WHERE table_name = '{BQ_TABLE_DENSE}'
    GROUP BY data_type
    ORDER BY column_count DESC
''').to_dataframe()
print(f'\nData types:')
print(type_inventory.to_string(index=False))
````

## 3. Sparse Events Table

The sparse table provides **multiple rows per entity** with timestamps, simulating time-series event data. Each entity has ~20 events spread over the past 90 days.

| Aspect | Value |
|--------|-------|
| **Entities** | Same 130,000 from dense table |
| **Events per entity** | 20 |
| **Total rows** | 2,600,000 |
| **Time range** | 90 days of history |
| **Event types** | purchase, view, click, search, return |
| **Metrics** | 5 continuous FLOAT64 columns |

````python
SPARSE_TABLE_ID = f'{BQ_PROJECT}.{BQ_DATASET}.{BQ_TABLE_SPARSE}'

print(f'Creating {SPARSE_TABLE_ID} (~2.6M rows)...')
print(f'This may take 30-60 seconds.\n')

sparse_query = f"""
CREATE OR REPLACE TABLE `{SPARSE_TABLE_ID}` AS
WITH
entities AS (
    SELECT entity_group, entity_id
    FROM `{DENSE_TABLE_ID}`
),
events_per_entity AS (
    SELECT entity_group, entity_id, event_num
    FROM entities
    CROSS JOIN UNNEST(GENERATE_ARRAY(1, 20)) AS event_num
)
SELECT
    entity_group,
    entity_id,
    TIMESTAMP_SUB(
        CURRENT_TIMESTAMP(),
        INTERVAL CAST(FLOOR(RAND() * 90 * 24 * 60) AS INT64) MINUTE
    ) AS event_timestamp,
    ['purchase', 'view', 'click', 'search', 'return'][ORDINAL(CAST(CEIL(RAND() * 5) AS INT64))] AS event_type,
    ROUND(RAND() * 100, 2) AS metric_1,
    ROUND(RAND() * 50, 2) AS metric_2,
    ROUND(RAND() * 200, 2) AS metric_3,
    ROUND(RAND() * 10, 2) AS metric_4,
    ROUND(RAND() * 1000, 2) AS metric_5,
    IF(RAND() > 0.9, 'positive', 'negative') AS label
FROM events_per_entity
"""

start = time.perf_counter()
bq.query(sparse_query).result()
elapsed = time.perf_counter() - start
print(f'Sparse table created in {elapsed:.1f}s')

info = bq.get_table(SPARSE_TABLE_ID)
print(f'  Rows:    {info.num_rows:,}')
print(f'  Columns: {len(info.schema)}')
````

### Verify Sparse Table

````python
print('=== Sparse Table Verification ===\n')

sparse_info = bq.get_table(SPARSE_TABLE_ID)
print(f'Rows: {sparse_info.num_rows:,}')

event_types = bq.query(f'''
    SELECT event_type, COUNT(*) AS n,
           ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(), 1) AS pct
    FROM `{SPARSE_TABLE_ID}`
    GROUP BY event_type ORDER BY n DESC
''').to_dataframe()
print(f'\nEvent type distribution:')
print(event_types.to_string(index=False))
````

## 4. Clustered Table

The BigQuery Feature Store notebooks use the [Storage Read API](https://cloud.google.com/bigquery/docs/reference/storage) with `row_restriction` to filter rows at the storage level. [Clustering](https://cloud.google.com/bigquery/docs/clustered-tables) on `entity_group, entity_id` physically sorts data so that a `row_restriction` like `entity_group = 'A' AND entity_id = '000001'` reads only the relevant storage blocks instead of scanning the full table.

This is the primary table used by the main notebook for all Storage Read API demonstrations.

````python
CLUSTERED_TABLE_ID = f'{BQ_PROJECT}.{BQ_DATASET}.{BQ_TABLE_CLUSTERED}'

print(f'Creating clustered table: {CLUSTERED_TABLE_ID}')
start = time.perf_counter()
bq.query(f'''
    CREATE OR REPLACE TABLE `{CLUSTERED_TABLE_ID}`
    CLUSTER BY entity_group, entity_id
    AS SELECT * FROM `{DENSE_TABLE_ID}`
''').result()
elapsed = time.perf_counter() - start
print(f'Created in {elapsed:.1f}s')

ct = bq.get_table(CLUSTERED_TABLE_ID)
print(f'  Rows:       {ct.num_rows:,}')
print(f'  Size:       {ct.num_bytes / 1024**2:.1f} MB')
print(f'  Clustering: {ct.clustering_fields}')
````

## 5. Embeddings Table

64-dimensional float vectors per entity for the vector similarity demo in the main notebook. Stored as `ARRAY<FLOAT64>` — the natural type for embedding vectors in BigQuery. Clustered by `entity_group, entity_id` for fast filtered reads.

````python
EMBEDDINGS_TABLE_ID = f'{BQ_PROJECT}.{BQ_DATASET}.{BQ_TABLE_EMBEDDINGS}'

print(f'Creating embeddings table: {EMBEDDINGS_TABLE_ID}')
start = time.perf_counter()
bq.query(f'''
    CREATE OR REPLACE TABLE `{EMBEDDINGS_TABLE_ID}`
    CLUSTER BY entity_group, entity_id
    AS
    SELECT
        entity_group,
        entity_id,
        ARRAY(
            SELECT RAND() - 0.5
            FROM UNNEST(GENERATE_ARRAY(1, 64))
        ) AS embedding
    FROM `{DENSE_TABLE_ID}`
''').result()
elapsed = time.perf_counter() - start
print(f'Created in {elapsed:.1f}s')

et = bq.get_table(EMBEDDINGS_TABLE_ID)
print(f'  Rows:       {et.num_rows:,}')
print(f'  Size:       {et.num_bytes / 1024**2:.1f} MB')
print(f'  Clustering: {et.clustering_fields}')

# Preview
sample = bq.query(f'''
    SELECT entity_group, entity_id, ARRAY_LENGTH(embedding) AS dims
    FROM `{EMBEDDINGS_TABLE_ID}` LIMIT 3
''').to_dataframe()
print(f'\nSample:')
print(sample.to_string(index=False))
````

## Cleanup

When you are done with the BigQuery Feature Store series, tear down the dataset with the `cleanup.py` script. It reads `cleanup.json` (written just above) so there's nothing to edit.

Unlike the other series, BigQuery has **no separate serving instance** — the serving layer is the dataset itself (Storage Read API straight from the offline tables). So the only resource to remove is the BigQuery dataset.

From this directory (`MLOps/Feature Store/BigQuery/`):

```bash
# Dry run — show what would be deleted (default, safe)
uv run python cleanup.py

# Delete the BigQuery dataset and all tables within it
uv run python cleanup.py --yes
```

The script is idempotent — an already-deleted dataset is reported, not errored.
