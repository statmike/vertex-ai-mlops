# Valkey Feature Store — Environment

This notebook creates the shared environment for the Valkey Feature Store series. It verifies/creates the BigQuery dataset, provisions a [Memorystore for Valkey](https://cloud.google.com/memorystore/docs/valkey/valkey-overview) instance, and sets up the [Pub/Sub](https://cloud.google.com/pubsub/docs/overview) topic used for BigQuery → Valkey sync.

**What this notebook creates:**
- BigQuery dataset `valkey_feature_store` with two tables (130K-entity dataset, 224 columns)
- A Private Service Connect (PSC) service connection policy (one-time networking prerequisite)
- A Memorystore for **Valkey** instance (Cluster Mode Disabled, Valkey 8.0)
- Pub/Sub topic + subscription for the sync bridge

**Compound entity key:** features are stored as Redis hashes with key pattern `features:{entity_group}:{entity_id}`.

---
## Why Memorystore for Valkey?

Google Cloud offers **two** managed in-memory stores. This series uses **Memorystore for Valkey**, but the patterns work on either. Here's the choice and why.

| | **Memorystore for Redis** | **Memorystore for Valkey** *(this series)* |
|---|---|---|
| **Engine** | Redis (up to 7.2 on Memorystore) | Valkey 7.2 / 8.0 / 9.0 |
| **Governance** | Redis Inc. (source-available SSPL/RSAL since 2024) | **Open source, Linux Foundation** (BSD) |
| **Architecture** | Single primary + optional replica | **Cluster-native** (shards × nodes); also a Cluster-Mode-Disabled option |
| **Scale-out** | Vertical (bigger instance, max 300 GB) | **Horizontal** — add shards to scale capacity and throughput |
| **Newer features** | Capped at 7.2 | Hash-field TTL (HEXPIRE), newer commands, ongoing OSS development |
| **Provisioning** | `gcloud redis` | `gcloud memorystore` (PSC-based networking) |
| **Client** | `redis.Redis` | `redis.Redis` (Cluster Disabled) or `RedisCluster` (Cluster Enabled) |

**Why we choose Valkey:**
1. **Open governance** — Valkey is the Linux Foundation fork created after Redis's 2024 license change; BSD-licensed, community-governed, no vendor lock-in.
2. **Horizontal scale** — cluster mode shards data across nodes, so you scale past a single machine's memory and throughput ceiling.
3. **Newer capabilities** — features like per-field hash TTL (`HEXPIRE`, Valkey 8.0+) that Memorystore for Redis (7.2) doesn't have.
4. **Drop-in compatibility** — Valkey is Redis-protocol compatible, so the same `redis-py` client and commands work; migration is mostly a provisioning change.

**Our setup choice — Cluster Mode Disabled:** we run a single-shard (Cluster Disabled) instance so the notebooks connect with the standard `redis.Redis` client and all multi-key operations (MULTI/EXEC, multi-key Lua, pipelining) work without hash-tag/slot constraints. We call out where **Cluster Mode Enabled** would differ.

> **🔄 Redis vs Valkey** callouts appear throughout the series wherever the Valkey approach differs from what you'd do on Memorystore for Redis.

````python
# Create tables if missing (full 224-column dataset matching all feature store series)
if 'dense_features' in missing:
    print('Creating dense_features table (130K rows, 224 columns)...')

    float_columns = []
    for i in range(6, 181):
        if i % 5 == 0:
            float_columns.append(f'    ROUND(RAND() * 1000, 4) AS feature_float_{i}')
        elif i % 5 == 1:
            float_columns.append(f'    ROUND(RAND() * 100 - 50, 4) AS feature_float_{i}')
        elif i % 5 == 2:
            float_columns.append(f'    ROUND(RAND() * 10, 4) AS feature_float_{i}')
        elif i % 5 == 3:
            float_columns.append(f'    ROUND(RAND() * 500 + 100, 4) AS feature_float_{i}')
        else:
            float_columns.append(f'    ROUND(RAND(), 6) AS feature_float_{i}')

    float_cols_sql = ',\n'.join(float_columns)

    dense_query = f"""
    CREATE OR REPLACE TABLE `{BQ_PROJECT}.{BQ_DATASET}.dense_features` AS
    WITH
    entity_groups AS (
        SELECT CHR(code) AS entity_group
        FROM UNNEST(GENERATE_ARRAY(65, 90)) AS code
    ),
    entity_ids AS (
        SELECT FORMAT('%06d', id) AS entity_id
        FROM UNNEST(GENERATE_ARRAY(1, 5000)) AS id
    ),
    entities AS (
        SELECT entity_group, entity_id
        FROM entity_groups CROSS JOIN entity_ids
    ),
    raw_features AS (
        SELECT
            entity_group,
            entity_id,
            CAST(FLOOR(RAND() * 1000) AS INT64) AS feature_int_1,
            CAST(FLOOR(RAND() * 10000) AS INT64) AS feature_int_2,
            CAST(FLOOR(RAND() * 100) AS INT64) AS feature_int_3,
            CAST(FLOOR(RAND() * 1000000) AS INT64) AS feature_int_4,
            CAST(MOD(ABS(FARM_FINGERPRINT(CONCAT(entity_group, entity_id))), 500) AS INT64) AS feature_int_5,
            ROUND(RAND() * 100, 4) AS feature_float_1,
            ROUND(RAND() * 1000, 4) AS feature_float_2,
            IF(RAND() > 0.05, ROUND(RAND() * 50 - 25, 4), NULL) AS feature_float_3,
            IF(RAND() > 0.05, ROUND(RAND() * 10, 4), NULL) AS feature_float_4,
            ROUND(RAND(), 6) AS feature_float_5,
            CAST(ROUND(RAND() * 99999, 2) AS NUMERIC) AS feature_numeric_1,
            IF(RAND() > 0.05, CAST(ROUND(RAND() * 9999999, 4) AS NUMERIC), NULL) AS feature_numeric_2,
            CAST(ROUND(RAND() * 100, 6) AS NUMERIC) AS feature_numeric_3,
            CAST(ROUND(RAND() * 999999999999, 2) AS BIGNUMERIC) AS feature_bignumeric_1,
            CAST(ROUND(RAND() * 99999999, 8) AS BIGNUMERIC) AS feature_bignumeric_2,
            CONCAT('cat_', CAST(CAST(FLOOR(RAND() * 20) AS INT64) AS STRING)) AS feature_string_1,
            IF(RAND() > 0.05, CONCAT('type_', CAST(CAST(FLOOR(RAND() * 50) AS INT64) AS STRING)), NULL) AS feature_string_2,
            ['alpha', 'beta', 'gamma', 'delta', 'epsilon', 'zeta', 'eta', 'theta'][ORDINAL(CAST(CEIL(RAND() * 8) AS INT64))] AS feature_string_3,
            CONCAT(entity_group, '_', ['north', 'south', 'east', 'west'][ORDINAL(CAST(CEIL(RAND() * 4) AS INT64))]) AS feature_string_4,
            FORMAT('%s-%06d', entity_group, CAST(FLOOR(RAND() * 999999) AS INT64)) AS feature_string_5,
            CAST(CONCAT('bytes_', entity_group, entity_id) AS BYTES) AS feature_bytes_1,
            CAST(FORMAT('%06d', CAST(FLOOR(RAND() * 999999) AS INT64)) AS BYTES) AS feature_bytes_2,
            RAND() > 0.5 AS feature_bool_1,
            RAND() > 0.7 AS feature_bool_2,
            MOD(CAST(FLOOR(RAND() * 10) AS INT64), 2) = 0 AS feature_bool_3,
            DATE_ADD(DATE '2020-01-01', INTERVAL CAST(FLOOR(RAND() * 1825) AS INT64) DAY) AS feature_date_1,
            IF(RAND() > 0.05, DATE_ADD(DATE '2022-06-01', INTERVAL CAST(FLOOR(RAND() * 730) AS INT64) DAY), NULL) AS feature_date_2,
            DATE_ADD(DATE '2023-01-01', INTERVAL CAST(FLOOR(RAND() * 365) AS INT64) DAY) AS feature_date_3,
            DATETIME_ADD(DATETIME '2023-01-01 00:00:00', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_datetime_1,
            DATETIME_ADD(DATETIME '2024-01-01 00:00:00', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_datetime_2,
            TIME_ADD(TIME '00:00:00', INTERVAL CAST(FLOOR(RAND() * 86400) AS INT64) SECOND) AS feature_time_1,
            TIME_ADD(TIME '00:00:00', INTERVAL CAST(FLOOR(RAND() * 86400) AS INT64) SECOND) AS feature_time_2,
            TIMESTAMP_ADD(TIMESTAMP '2023-01-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_timestamp_1,
            TIMESTAMP_ADD(TIMESTAMP '2024-01-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 525600) AS INT64) MINUTE) AS feature_timestamp_2,
            TIMESTAMP_ADD(TIMESTAMP '2024-06-01 00:00:00 UTC', INTERVAL CAST(FLOOR(RAND() * 262800) AS INT64) MINUTE) AS feature_timestamp_3,
            ST_GEOGPOINT(ROUND(-180 + RAND() * 360, 6), ROUND(-90 + RAND() * 180, 6)) AS feature_geo_1,
            ST_GEOGPOINT(ROUND(-122.5 + RAND() * 5, 6), ROUND(37.0 + RAND() * 3, 6)) AS feature_geo_2,
            JSON_OBJECT('score', ROUND(RAND() * 100, 2), 'label', CONCAT('class_', CAST(CAST(FLOOR(RAND() * 10) AS INT64) AS STRING)), 'active', RAND() > 0.5) AS feature_json_1,
            JSON_OBJECT('x', ROUND(RAND() * 10, 4), 'y', ROUND(RAND() * 10, 4), 'tag', entity_group) AS feature_json_2,
            [CAST(FLOOR(RAND() * 100) AS INT64), CAST(FLOOR(RAND() * 100) AS INT64), CAST(FLOOR(RAND() * 100) AS INT64)] AS feature_array_int_1,
            [CAST(FLOOR(RAND() * 1000) AS INT64), CAST(FLOOR(RAND() * 1000) AS INT64)] AS feature_array_int_2,
            [['red', 'blue', 'green', 'yellow'][ORDINAL(CAST(CEIL(RAND() * 4) AS INT64))], ['circle', 'square', 'triangle'][ORDINAL(CAST(CEIL(RAND() * 3) AS INT64))]] AS feature_array_string_1,
            STRUCT(ROUND(RAND() * 100, 2) AS value, CONCAT('unit_', CAST(CAST(FLOOR(RAND() * 5) AS INT64) AS STRING)) AS unit, RAND() > 0.5 AS is_valid) AS feature_struct_1,
            STRUCT(CAST(FLOOR(RAND() * 1000) AS INT64) AS count, ROUND(RAND(), 4) AS ratio) AS feature_struct_2,
            [STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name), STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name), STRUCT(ROUND(RAND() * 10, 2) AS weight, CONCAT('item_', CAST(CAST(FLOOR(RAND() * 100) AS INT64) AS STRING)) AS name)] AS feature_array_struct_1,
            RANGE(DATE_ADD(DATE '2020-01-01', INTERVAL CAST(FLOOR(RAND() * 1000) AS INT64) DAY), DATE_ADD(DATE '2023-01-01', INTERVAL CAST(FLOOR(RAND() * 730) AS INT64) DAY)) AS feature_range_date_1,
        {float_cols_sql}
        FROM entities
    )
    SELECT *,
        CASE WHEN (
            0.4 * feature_int_1 / 999.0 +
            0.3 * feature_float_1 / 100.0 +
            0.15 * (COALESCE(feature_float_3, 0.0) + 25.0) / 50.0 +
            0.15 * CAST(feature_bool_1 AS INT64)
            + (RAND() - 0.5) * 0.3
        ) > 0.6 THEN 'positive' ELSE 'negative' END AS label
    FROM raw_features
    """

    bq.query(dense_query).result()
    print('  dense_features created.')

if 'sparse_events' in missing:
    print('Creating sparse_events table (~2.6M rows)...')

    sparse_query = f"""
    CREATE OR REPLACE TABLE `{BQ_PROJECT}.{BQ_DATASET}.sparse_events` AS
    WITH
    entities AS (
        SELECT entity_group, entity_id
        FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
    ),
    events_per_entity AS (
        SELECT entity_group, entity_id, event_num
        FROM entities CROSS JOIN UNNEST(GENERATE_ARRAY(1, 20)) AS event_num
    )
    SELECT
        entity_group, entity_id,
        TIMESTAMP_SUB(CURRENT_TIMESTAMP(), INTERVAL CAST(FLOOR(RAND() * 90 * 24 * 60) AS INT64) MINUTE) AS event_timestamp,
        ['purchase', 'view', 'click', 'search', 'return'][ORDINAL(CAST(CEIL(RAND() * 5) AS INT64))] AS event_type,
        ROUND(RAND() * 100, 2) AS metric_1,
        ROUND(RAND() * 50, 2) AS metric_2,
        ROUND(RAND() * 200, 2) AS metric_3,
        ROUND(RAND() * 10, 2) AS metric_4,
        ROUND(RAND() * 1000, 2) AS metric_5,
        IF(RAND() > 0.9, 'positive', 'negative') AS label
    FROM events_per_entity
    """

    bq.query(sparse_query).result()
    print('  sparse_events created.')

if not missing:
    print('Both tables already exist — no action taken.')
````

## Memorystore for Valkey

### Architecture

[Memorystore for Valkey](https://cloud.google.com/memorystore/docs/valkey/valkey-overview) is Google Cloud's managed Valkey service. Unlike Memorystore for Redis (primary + replica), Valkey is **cluster-native**:

```
Instance (Cluster Mode Disabled, this series)
└── 1 shard
    ├── Primary node   ← reads + writes (single endpoint)
    └── Replica node   ← automatic failover (HA)

Instance (Cluster Mode Enabled, for scale-out)
├── Shard 1 (primary + replicas)
├── Shard 2 (primary + replicas)   ← data sharded across nodes
└── ...                              (discovery endpoint, RedisCluster client)
```

| Concept | Description |
|---|---|
| **Mode** | `cluster-disabled` (single endpoint, this series) or `cluster` (sharded scale-out) |
| **Node type** | Machine size per node (`standard-small` here; up to `highmem-*` for production) |
| **Shards** | Horizontal scale unit (1 here; add shards in cluster mode for more capacity) |
| **Replicas** | Per-shard replicas for HA (1 here → automatic failover) |
| **Networking** | Private Service Connect (PSC) automation — requires a service connection policy |

> **🔄 Redis vs Valkey:** Memorystore for Redis uses direct VPC peering and a simple host/port; Memorystore for Valkey uses **PSC automation**, so the one-time prerequisite below (service connection policy) has no Redis equivalent.

### Step 1: Service Connection Policy (PSC prerequisite)

````python
# Memorystore for Valkey connects via Private Service Connect (PSC) automation.
# Prerequisite: a service connection policy for the gcp-memorystore service class
# on your VPC + region. This is a one-time setup per (network, region).
import subprocess, json

def scp_exists(name):
    r = subprocess.run(
        ['gcloud', 'network-connectivity', 'service-connection-policies', 'describe',
         name, '--region', REGION, '--project', PROJECT_ID, '--format', 'value(name)'],
        capture_output=True, text=True)
    return r.returncode == 0

if scp_exists(SCP_NAME):
    print(f'Service connection policy exists: {SCP_NAME}')
else:
    print(f'Creating service connection policy {SCP_NAME}...')
    subnet = f'projects/{PROJECT_ID}/regions/{REGION}/subnetworks/{VPC_NETWORK}'
    r = subprocess.run(
        ['gcloud', 'network-connectivity', 'service-connection-policies', 'create',
         SCP_NAME,
         '--network', VPC_NETWORK,
         '--service-class', 'gcp-memorystore',
         '--region', REGION,
         '--subnets', subnet,
         '--project', PROJECT_ID],
        capture_output=True, text=True)
    if r.returncode == 0:
        print(f'  Created: {SCP_NAME}')
    else:
        print(f'  Note: {r.stderr.strip()}')
````

### Step 2: Create the Valkey Instance

````python
# Create the Memorystore for Valkey instance (Cluster Mode Disabled).
# This takes several minutes. Cluster-disabled gives a single primary endpoint,
# so we connect with the standard redis.Redis client (no cluster awareness needed).
import subprocess, json, time

def get_instance():
    r = subprocess.run(
        ['gcloud', 'memorystore', 'instances', 'describe', VALKEY_INSTANCE,
         '--location', REGION, '--project', PROJECT_ID, '--format', 'json'],
        capture_output=True, text=True)
    return json.loads(r.stdout) if r.returncode == 0 else None

inst = get_instance()
if inst:
    print(f'Valkey instance exists: {VALKEY_INSTANCE} (state: {inst.get("state")})')
else:
    print(f'Creating Valkey instance {VALKEY_INSTANCE} (3-5 min)...')
    r = subprocess.run(
        ['gcloud', 'memorystore', 'instances', 'create', VALKEY_INSTANCE,
         '--location', REGION,
         '--node-type', VALKEY_NODE_TYPE,
         '--shard-count', '1',
         '--replica-count', '1',          # 1 replica = HA with automatic failover
         '--mode', VALKEY_MODE,
         '--engine-version', VALKEY_ENGINE_VERSION,
         '--transit-encryption-mode', 'server-authentication',
         '--psc-auto-connections', f'network=projects/{PROJECT_ID}/global/networks/{VPC_NETWORK},projectId={PROJECT_ID}',
         '--project', PROJECT_ID],
        capture_output=True, text=True)
    if r.returncode == 0:
        print('  Created.')
    else:
        print(f'  {r.stderr.strip()}')
    inst = get_instance()

if inst:
    print(f'  Engine: Valkey {inst.get("engineVersion", VALKEY_ENGINE_VERSION)}')
    print(f'  Node type: {inst.get("nodeType")}')
    print(f'  Mode: {inst.get("mode")}')
````

### Step 3: Discover Endpoint + TLS Certificate

````python
# Discover the connection endpoint. For a Cluster-Mode-Disabled instance the
# describe output lists a PRIMARY endpoint (IP + port) under endpoints/connections.
import subprocess, json, os

inst = get_instance()
if inst is None:
    raise RuntimeError(
        'Valkey instance not found — the create step above likely failed. '
        'Re-run the previous cell and check its error output before continuing.')

def find_primary_endpoint(inst):
    # endpoints is a list; each has connections with a pscAutoConnection (ip, port)
    for ep in inst.get('endpoints', []):
        for conn in ep.get('connections', []):
            psc = conn.get('pscAutoConnection', {})
            if psc.get('connectionType', '').endswith('PRIMARY') or 'ipAddress' in psc:
                if psc.get('ipAddress'):
                    return psc['ipAddress'], int(psc.get('port', 6379))
    # Fallback: some versions expose discoveryEndpoints
    for de in inst.get('discoveryEndpoints', []):
        if de.get('address'):
            return de['address'], int(de.get('port', 6379))
    return None, None

VALKEY_HOST, VALKEY_PORT = find_primary_endpoint(inst)
print(f'Valkey endpoint: {VALKEY_HOST}:{VALKEY_PORT}')

# Retrieve the server CA certificate for TLS (server-authentication mode)
CERT_DIR = os.path.join(os.path.expanduser('~'), '.valkey-certs')
CERT_PATH = os.path.join(CERT_DIR, f'{VALKEY_INSTANCE}-ca.pem')
os.makedirs(CERT_DIR, exist_ok=True)

r = subprocess.run(
    ['gcloud', 'memorystore', 'instances', 'get-certificate-authority',
     VALKEY_INSTANCE, '--location', REGION, '--project', PROJECT_ID, '--format', 'json'],
    capture_output=True, text=True)
if r.returncode == 0:
    ca = json.loads(r.stdout)
    pem = ''
    # managedServerCa.caCerts[].certificates[] holds PEM strings
    for cert in ca.get('managedServerCa', {}).get('caCerts', []):
        for c in cert.get('certificates', []):
            pem += c + '\n'
    if pem:
        with open(CERT_PATH, 'w') as f:
            f.write(pem)
        print(f'CA certificate saved: {CERT_PATH}')
    else:
        CERT_PATH = None
        print('No CA cert found in output; will connect without ssl_ca_certs.')
else:
    CERT_PATH = None
    print(f'Certificate retrieval note: {r.stderr.strip()}')
````

### Step 4: Connect and Verify

> **🔄 Redis vs Valkey:** with **Cluster Mode Disabled** we use the standard `redis.Redis` client — identical to Memorystore for Redis. With **Cluster Mode Enabled** you'd use `redis.cluster.RedisCluster` against the discovery endpoint, and multi-key ops would require hash-tagged keys to keep related keys in the same slot.

````python
import redis

if VALKEY_HOST:
    # Cluster Mode Disabled -> standard redis.Redis client (NOT RedisCluster).
    # Two-client pattern: decode_responses=True for strings, False for bytes (embeddings).
    connect_kwargs = dict(host=VALKEY_HOST, port=VALKEY_PORT)
    if CERT_PATH:
        # Memorystore's server cert is issued for a DNS name, not the PSC IP we
        # connect to. Verify the cert chain against the CA, but skip the hostname
        # match (ssl_check_hostname=False) — otherwise TLS fails with an IP mismatch.
        connect_kwargs.update(ssl=True, ssl_ca_certs=CERT_PATH, ssl_check_hostname=False)

    decode_client = redis.Redis(decode_responses=True, **connect_kwargs)
    bytes_client = redis.Redis(decode_responses=False, **connect_kwargs)

    info = decode_client.info('server')
    # Valkey reports its own version in valkey_version; redis_version is just the
    # Redis-protocol compatibility level (7.2.4 even on Valkey 8.x).
    print('Connected to Valkey:')
    print(f'  Valkey version:   {info.get("valkey_version", "n/a")}')
    print(f'  Redis-compat:     {info.get("redis_version")}')
    print(f'  Server:           {info.get("server_name", "valkey")}')

    # Quick latency check
    import time
    decode_client.hset('test:ping', mapping={'value': 'pong'})
    lats = []
    for _ in range(100):
        s = time.perf_counter(); decode_client.hgetall('test:ping'); lats.append((time.perf_counter()-s)*1000)
    print(f'  Point read median: {sorted(lats)[50]:.3f} ms')
    decode_client.delete('test:ping')
else:
    print('Valkey endpoint not available. For local dev: docker run -p 6379:6379 valkey/valkey:8')
````

## Pub/Sub Topic and Subscription

Create the Pub/Sub infrastructure for the BigQuery → Valkey sync bridge.

```
BigQuery (continuous query)
    │
    └─ INSERT INTO pubsub.topic()
           │
           └─ Pub/Sub topic: valkey-feature-sync
                  │
                  └─ Subscription: valkey-feature-sync-sub
                         │
                         └─ Python subscriber → HSET to Valkey
```

````python
topic_path = publisher.topic_path(PROJECT_ID, PUBSUB_TOPIC)
subscription_path = subscriber.subscription_path(PROJECT_ID, PUBSUB_SUBSCRIPTION)

# Create topic
try:
    topic = publisher.create_topic(name=topic_path)
    print(f'Created topic: {topic.name}')
except Exception as e:
    if 'ALREADY_EXISTS' in str(e):
        print(f'Topic exists: {topic_path}')
    else:
        print(f'Topic: {e}')

# Create subscription
try:
    sub = subscriber.create_subscription(
        name=subscription_path,
        topic=topic_path,
        ack_deadline_seconds=60,
    )
    print(f'Created subscription: {sub.name}')
except Exception as e:
    if 'ALREADY_EXISTS' in str(e):
        print(f'Subscription exists: {subscription_path}')
    else:
        print(f'Subscription: {e}')
````

````python
# Verify Pub/Sub with a test message
import json

test_message = json.dumps({'entity_group': 'TEST', 'entity_id': '000001', 'feature_float_1': 42.0})
future = publisher.publish(topic_path, test_message.encode('utf-8'))
msg_id = future.result()
print(f'Published test message: {msg_id}')

# Pull the message back
import time
time.sleep(2)  # Wait for delivery

response = subscriber.pull(subscription=subscription_path, max_messages=10)
for msg in response.received_messages:
    data = json.loads(msg.message.data.decode('utf-8'))
    print(f'Received: {data}')
    subscriber.acknowledge(subscription=subscription_path, ack_ids=[msg.ack_id])

if not response.received_messages:
    print('No messages received (may need a moment to propagate).')
````

## Data Structure for Redis/Valkey

Redis/Valkey stores features as **hashes** — one hash per entity, with feature names as fields.

```
Key: features:A:000001
Fields:
  feature_int_1    = "499"
  feature_float_1  = "50.1095"
  feature_string_1 = "cat_15"
  feature_bool_1   = "0"
  label            = "negative"
```

| Redis Concept | Bigtable Equivalent | Spanner Equivalent |
|---|---|---|
| Hash key (`features:A:000001`) | Row key (`A#000001`) | Primary key `('A', '000001')` |
| Hash field (`feature_float_1`) | Column qualifier | Column name |
| Hash value (`"50.1095"`) | Cell value (bytes) | Column value (typed) |
| HSET | Mutation | INSERT/UPDATE |
| HGETALL | ReadRow | SELECT * |
