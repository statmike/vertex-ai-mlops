# Valkey Feature Store — Production Deployment and Capstone (Real-Time Trading)

Terraform deployment + real-time trading feature store that demonstrates everything unique about Valkey: sub-ms reads, HNSW instrument matching, atomic P&L tracking, TTL for expiring quotes, and sorted sets for order books.

**What you'll learn:**
- Terraform: `google_redis_instance` (Standard tier, Redis 7.2+)
- Real-time trading feature store:
  - Instrument features as hashes with sub-ms reads
  - Market quotes with TTL (5-second expiration)
  - HNSW index for similar instrument discovery
  - Atomic P&L tracking with `HINCRBYFLOAT`
  - Sorted sets for top movers and order book depth
- Two-stage instrument recommendation
- FastAPI endpoint with load testing

**Why trading?** It's the canonical sub-ms use case. Stale prices are dangerous, position lookups must be instant, and similar-instrument discovery powers hedging and opportunity detection.

**Prerequisites:** Run [Environment](./Valkey%20Feature%20Store%20-%20Environment.ipynb) first.

## 2. Capstone: Real-Time Trading Feature Store

### Data Model

```
instrument:AAPL          ← Hash: price, volume, sector, market_cap, ...
trader:T001              ← Hash: position_count, realized_pnl, risk_score, ...
quote:AAPL:bid           ← Hash: price, size (EXPIRE 5 seconds)
quote:AAPL:ask           ← Hash: price, size (EXPIRE 5 seconds)
top_gainers              ← Sorted Set: instruments ranked by % change
emb:instrument:AAPL      ← Hash: factor exposure embedding (for HNSW)
```

### Load Trading Data

````python
if REDIS_AVAILABLE:
    np.random.seed(42)
    random.seed(42)

    sectors = ['tech', 'finance', 'healthcare', 'energy', 'consumer']
    tickers = []
    for sector_idx, sector in enumerate(sectors):
        for i in range(200):  # 200 instruments per sector = 1000 total
            ticker = f'{sector[:3].upper()}{i:04d}'
            tickers.append(ticker)

    # Load instrument features
    pipe = decode_client.pipeline(transaction=False)
    for i, ticker in enumerate(tickers):
        sector = sectors[i // 200]
        price = round(random.uniform(5, 500), 2)
        volume = random.randint(10000, 10000000)
        market_cap = round(price * volume * random.uniform(10, 100), 0)
        pct_change = round(random.uniform(-5, 5), 2)

        pipe.hset(f'instrument:{ticker}', mapping={
            'ticker': ticker,
            'sector': sector,
            'price': str(price),
            'volume': str(volume),
            'market_cap': str(market_cap),
            'pct_change': str(pct_change),
            'beta': str(round(random.uniform(0.5, 2.0), 3)),
            'pe_ratio': str(round(random.uniform(5, 50), 1)),
        })

        # Add to top_gainers sorted set
        pipe.zadd('top_gainers', {ticker: pct_change})

        if (i + 1) % 500 == 0:
            pipe.execute()
            pipe = decode_client.pipeline(transaction=False)
    pipe.execute()
    print(f'Loaded {len(tickers)} instrument features')

    # Load trader profiles
    n_traders = 100
    pipe = decode_client.pipeline(transaction=False)
    for i in range(n_traders):
        tid = f'T{i:04d}'
        pipe.hset(f'trader:{tid}', mapping={
            'trader_id': tid,
            'position_count': str(random.randint(0, 50)),
            'realized_pnl': str(round(random.uniform(-50000, 100000), 2)),
            'risk_score': str(round(random.uniform(0, 1), 3)),
            'trade_count': str(random.randint(0, 1000)),
            'preferred_sector': random.choice(sectors),
        })
    pipe.execute()
    print(f'Loaded {n_traders} trader profiles')

    # Load factor exposure embeddings for HNSW
    pipe = bytes_client.pipeline(transaction=False)
    for i, ticker in enumerate(tickers):
        emb = np.random.randn(EMBEDDING_DIM).astype(np.float32)
        emb = emb / np.linalg.norm(emb)
        pipe.hset(f'emb:instrument:{ticker}'.encode(), mapping={
            b'embedding': emb.tobytes(),
            b'ticker': ticker.encode(),
            b'sector': sectors[i // 200].encode(),
        })
        if (i + 1) % 500 == 0:
            pipe.execute()
            pipe = bytes_client.pipeline(transaction=False)
    pipe.execute()
    print(f'Loaded {len(tickers)} factor exposure embeddings ({EMBEDDING_DIM}-dim)')
````

### Create HNSW Index for Instrument Matching

````python
if REDIS_AVAILABLE:
    try:
        decode_client.execute_command('FT.DROPINDEX', 'idx:instruments')
    except Exception:
        pass

    decode_client.execute_command(
        'FT.CREATE', 'idx:instruments',
        'ON', 'HASH',
        'PREFIX', '1', 'emb:instrument:',
        'SCHEMA',
        'embedding', 'VECTOR', 'HNSW', '10',
        'TYPE', 'FLOAT32',
        'DIM', str(EMBEDDING_DIM),
        'DISTANCE_METRIC', 'COSINE',
        'M', '16',
        'EF_CONSTRUCTION', '200',
        'sector', 'TAG',
    )
    time.sleep(2)
    print('HNSW index created for instrument matching.')
````

## 3. Market Quotes with TTL

Market quotes expire in 5 seconds — stale prices are dangerous in trading.

````python
if REDIS_AVAILABLE:
    # Simulate quote updates
    def publish_quote(ticker, side, price, size, ttl_seconds=5):
        key = f'quote:{ticker}:{side}'
        decode_client.hset(key, mapping={'price': str(price), 'size': str(size)})
        decode_client.expire(key, ttl_seconds)

    # Publish some quotes
    publish_quote('TEC0001', 'bid', 195.40, 1000)
    publish_quote('TEC0001', 'ask', 195.45, 800)
    publish_quote('FIN0001', 'bid', 42.10, 5000)
    publish_quote('FIN0001', 'ask', 42.15, 3000)

    print('Published quotes (5-second TTL):\n')
    for ticker in ['TEC0001', 'FIN0001']:
        bid = decode_client.hgetall(f'quote:{ticker}:bid')
        ask = decode_client.hgetall(f'quote:{ticker}:ask')
        ttl = decode_client.ttl(f'quote:{ticker}:bid')
        spread = float(ask['price']) - float(bid['price'])
        print(f'  {ticker}: bid={bid["price"]} x {bid["size"]}  ask={ask["price"]} x {ask["size"]}  spread={spread:.2f}  TTL={ttl}s')

    print(f'\nWaiting 6 seconds for quotes to expire...')
    time.sleep(6)

    for ticker in ['TEC0001', 'FIN0001']:
        exists = decode_client.exists(f'quote:{ticker}:bid')
        print(f'  quote:{ticker}:bid exists: {bool(exists)} (expired)')
````

## 4. Atomic P&L Tracking

When a trade executes, update the trader's P&L atomically.

````python
if REDIS_AVAILABLE:
    trader_key = 'trader:T0001'
    before = decode_client.hgetall(trader_key)
    print(f'Before trades:')
    print(f'  P&L:         ${float(before["realized_pnl"]):,.2f}')
    print(f'  Trade count: {before["trade_count"]}')

    # Simulate 3 trades
    trades = [
        ('TEC0001', 100, 195.40, 196.50),   # Buy 100 @ 195.40, sell @ 196.50 → +110
        ('FIN0001', 500, 42.10, 41.80),     # Buy 500 @ 42.10, sell @ 41.80 → -150
        ('HEA0001', 200, 88.00, 90.25),     # Buy 200 @ 88.00, sell @ 90.25 → +450
    ]

    for ticker, qty, buy, sell in trades:
        pnl = (sell - buy) * qty

        # Atomic updates
        start = time.perf_counter()
        pipe = decode_client.pipeline(transaction=True)
        pipe.hincrbyfloat(trader_key, 'realized_pnl', pnl)
        pipe.hincrby(trader_key, 'trade_count', 1)
        pipe.hincrby(trader_key, 'position_count', 0)  # Net zero (closed position)
        results = pipe.execute()
        trade_ms = (time.perf_counter() - start) * 1000

        print(f'\n  Trade: {qty} {ticker} @ {buy}→{sell} = ${pnl:+,.2f} ({trade_ms:.2f}ms)')
        print(f'    New P&L: ${float(results[0]):,.2f}, trades: {results[1]}')

    after = decode_client.hgetall(trader_key)
    print(f'\nAfter trades:')
    print(f'  P&L:         ${float(after["realized_pnl"]):,.2f}')
    print(f'  Trade count: {after["trade_count"]}')
````

## 5. Sorted Sets — Top Movers and Order Book

````python
if REDIS_AVAILABLE:
    # Top gainers (highest % change)
    top_gainers = decode_client.zrevrangebyscore('top_gainers', '+inf', '-inf', start=0, num=10, withscores=True)
    print('Top 10 Gainers:')
    for rank, (ticker, pct) in enumerate(top_gainers, 1):
        instrument = decode_client.hgetall(f'instrument:{ticker}')
        print(f'  {rank:>2}. {ticker:<10s} {pct:>+6.2f}%  ${float(instrument["price"]):>8.2f}  {instrument["sector"]}')

    # Top losers (lowest % change)
    top_losers = decode_client.zrangebyscore('top_gainers', '-inf', '+inf', start=0, num=5, withscores=True)
    print(f'\nTop 5 Losers:')
    for rank, (ticker, pct) in enumerate(top_losers, 1):
        print(f'  {rank:>2}. {ticker:<10s} {pct:>+6.2f}%')
````

## 6. HNSW Instrument Matching

Find instruments with similar factor exposure — used for hedging and opportunity detection.

````python
if REDIS_AVAILABLE:
    # Get embedding for a query instrument
    query_ticker = 'TEC0050'
    query_bytes = bytes_client.hget(f'emb:instrument:{query_ticker}'.encode(), b'embedding')

    if query_bytes:
        K = 10
        start = time.perf_counter()
        results = bytes_client.execute_command(
            'FT.SEARCH', b'idx:instruments',
            b'*=>[KNN ' + str(K).encode() + b' @embedding $q AS distance]',
            b'PARAMS', b'2', b'q', query_bytes,
            b'RETURN', b'3', b'ticker', b'sector', b'distance',
            b'DIALECT', b'2',
        )
        search_ms = (time.perf_counter() - start) * 1000

        # Parse + sort by distance client-side (KNN order not guaranteed without SORTBY,
        # which Memorystore's managed search rejects).
        n_results = results[0]
        matches = []
        for i in range(1, min(n_results * 2 + 1, len(results)), 2):
            fields = dict(zip(results[i + 1][::2], results[i + 1][1::2]))
            matches.append((
                fields.get(b'ticker', b'?').decode(),
                fields.get(b'sector', b'?').decode(),
                float(fields.get(b'distance', b'inf')),
            ))
        matches.sort(key=lambda x: x[2])
        print(f'Instruments similar to {query_ticker} (HNSW, {search_ms:.1f}ms):\n')
        for ticker, sector, dist in matches:
            price = decode_client.hget(f'instrument:{ticker}', 'price') or '?'
            print(f'  {ticker:<10s} {sector:<12s} price=${price:>8s}  distance={dist:.4f}')
    else:
        print(f'Embedding not found for {query_ticker}')
````

## 7. Two-Stage Instrument Recommendation

1. **Stage 1**: HNSW retrieval — find 50 instruments with similar factor exposure (<5ms)
2. **Stage 2**: Read instrument features + apply risk scoring → top 10

````python
if REDIS_AVAILABLE:
    def recommend_instruments(query_ticker, trader_id, top_k=10):
        timings = {}

        # Get query embedding
        q_bytes = bytes_client.hget(f'emb:instrument:{query_ticker}'.encode(), b'embedding')
        if not q_bytes:
            return None, timings

        # Stage 1: HNSW retrieval
        t0 = time.perf_counter()
        results = bytes_client.execute_command(
            'FT.SEARCH', b'idx:instruments',
            b'*=>[KNN 50 @embedding $q AS distance]',
            b'PARAMS', b'2', b'q', q_bytes,
            b'RETURN', b'2', b'ticker', b'distance',
            b'DIALECT', b'2',
        )
        timings['hnsw'] = (time.perf_counter() - t0) * 1000

        # Parse candidates
        candidates = []
        n = results[0]
        for i in range(1, min(n * 2 + 1, len(results)), 2):
            fields = dict(zip(results[i + 1][::2], results[i + 1][1::2]))
            ticker = fields.get(b'ticker', b'?').decode()
            dist = float(fields.get(b'distance', b'999'))
            if ticker != query_ticker:
                candidates.append((ticker, dist))
        # Sort by distance client-side (order not guaranteed without SORTBY)
        candidates.sort(key=lambda x: x[1])

        # Stage 2: Read features + score
        t1 = time.perf_counter()
        trader = decode_client.hgetall(f'trader:{trader_id}')
        pref_sector = trader.get('preferred_sector', 'tech')

        scored = []
        pipe = decode_client.pipeline(transaction=False)
        for ticker, dist in candidates[:30]:
            pipe.hgetall(f'instrument:{ticker}')
        instrument_data = pipe.execute()

        for (ticker, dist), inst in zip(candidates[:30], instrument_data):
            if not inst:
                continue
            sector_match = 1.0 if inst.get('sector') == pref_sector else 0.0
            pct = float(inst.get('pct_change', '0'))
            beta = float(inst.get('beta', '1'))

            score = (
                0.3 * (1.0 - dist) +    # Similarity
                0.2 * sector_match +      # Sector preference
                0.2 * max(0, pct) / 5.0 + # Positive momentum
                0.3 * (1.0 / beta)        # Lower beta preferred
            )
            scored.append({
                'ticker': ticker,
                'sector': inst.get('sector', '?'),
                'price': inst.get('price', '?'),
                'pct_change': pct,
                'similarity': 1.0 - dist,
                'score': score,
            })

        scored.sort(key=lambda x: -x['score'])
        timings['scoring'] = (time.perf_counter() - t1) * 1000
        timings['total'] = sum(timings.values())

        return scored[:top_k], timings

    recs, timings = recommend_instruments('TEC0050', 'T0001')

    print(f'Instrument recommendations for trader T0001 (query: TEC0050):\n')
    print(f'  HNSW retrieval:  {timings["hnsw"]:.1f}ms')
    print(f'  Feature + score: {timings["scoring"]:.1f}ms')
    print(f'  Total:           {timings["total"]:.1f}ms\n')

    print(f'{"Ticker":>10s} {"Sector":>12s} {"Price":>8s} {"Chg%":>6s} {"Sim":>6s} {"Score":>7s}')
    for r in recs:
        print(f'{r["ticker"]:>10s} {r["sector"]:>12s} ${float(r["price"]):>6.2f} {r["pct_change"]:>+5.2f}% {r["similarity"]:>5.3f} {r["score"]:>6.3f}')
````

## 8. Load Test — Recommendation Pipeline

````python
if REDIS_AVAILABLE:
    from concurrent.futures import ThreadPoolExecutor, as_completed

    def single_recommendation(args):
        ticker, trader = args
        start = time.perf_counter()
        recs, _ = recommend_instruments(ticker, trader, top_k=5)
        return (time.perf_counter() - start) * 1000

    test_requests = [
        (f'{sectors[i % 5][:3].upper()}{(i * 7) % 200:04d}', f'T{i % 100:04d}')
        for i in range(200)
    ]

    # Sequential
    seq_lats = []
    start = time.perf_counter()
    for req in test_requests[:50]:
        seq_lats.append(single_recommendation(req))
    seq_wall = (time.perf_counter() - start) * 1000

    # Concurrent
    conc_lats = []
    start = time.perf_counter()
    with ThreadPoolExecutor(max_workers=20) as executor:
        futures = [executor.submit(single_recommendation, r) for r in test_requests[:50]]
        conc_lats = [f.result() for f in as_completed(futures)]
    conc_wall = (time.perf_counter() - start) * 1000

    print(f'Recommendation pipeline load test (50 requests):\n')
    print(f'{"Mode":<15s} {"p50":>8s} {"p95":>8s} {"p99":>8s} {"Wall":>10s} {"RPS":>8s}')
    for name, lats, wall in [('Sequential', seq_lats, seq_wall), ('Concurrent (20)', conc_lats, conc_wall)]:
        s = sorted(lats)
        rps = 50 / (wall / 1000)
        print(f'{name:<15s} {s[25]:>6.1f}ms {s[47]:>6.1f}ms {s[49]:>6.1f}ms {wall:>8.0f}ms {rps:>6.0f}')
````

## 11. Security — AUTH, TLS, IAM, and Network Isolation

Three layers: **network** (private IP only — no public endpoint), **transport** (TLS, used throughout this series via the CA cert), and **auth** (AUTH password + IAM for control-plane management). Least-privilege IAM separates serving, sync, and admin roles.

````python
iam_and_auth = f"""
# --- Authentication ---
# Memorystore supports AUTH (a password) and in-transit TLS (we use TLS in this
# series via the CA cert). Enable AUTH on the instance, then pass the password:
#
#   redis.Redis(host=..., port=..., password=AUTH_STRING,
#               ssl=True, ssl_ca_certs=CERT_PATH)
#
# Retrieve the AUTH string:
gcloud redis instances get-auth-string {REDIS_INSTANCE} --region {REGION}

# --- IAM (control-plane access, not data-plane) ---
# IAM governs who can manage the instance; AUTH+TLS govern data access.
# Least-privilege roles for a feature store:

# Serving service account: read instance metadata + connect (no admin)
gcloud projects add-iam-policy-binding {PROJECT_ID} \\
    --member serviceAccount:serving-sa@{PROJECT_ID}.iam.gserviceaccount.com \\
    --role roles/redis.viewer

# Sync/pipeline service account: same connect rights; data writes via AUTH
gcloud projects add-iam-policy-binding {PROJECT_ID} \\
    --member serviceAccount:sync-sa@{PROJECT_ID}.iam.gserviceaccount.com \\
    --role roles/redis.viewer

# Platform/admin: full management (create, scale, delete)
#   roles/redis.admin   (restrict to platform team only)

# --- Network ---
# Memorystore is private-IP only by default. Clients reach it via:
#   - VPC (same network), or
#   - Serverless VPC Access connector (Cloud Run / Cloud Functions), or
#   - VPC peering / PSC for cross-project.
# There is no public endpoint — network isolation is the first security layer.
"""
print(iam_and_auth)
````

## 14. Production Readiness Checklist

A consolidated checklist for running a Valkey feature store in production — synthesizing the whole series.

**Infrastructure**
- [ ] Standard HA tier (primary + replica) for production serving
- [ ] Instance sized with headroom above peak `used_memory` (NB4 §8)
- [ ] `maxmemory-policy = volatile-lru` set at instance level (NB4 §5)
- [ ] Private IP + VPC connector for Cloud Run / serverless clients
- [ ] TLS enabled (CA cert distributed to clients); AUTH enabled

**Data & Sync**
- [ ] BigQuery is the source of truth; Valkey is rebuildable from sync (NB3)
- [ ] Sync pipeline deployed (continuous query → Pub/Sub → subscriber, NB3 §3–4)
- [ ] TTL-on-sync for automatic staleness bounds (NB3 §5)
- [ ] Dead-letter topic for failed sync messages (NB3 §10)

**Serving**
- [ ] Connection pool created once; socket timeouts set (NB6 §7)
- [ ] Retry with backoff; explicit miss/timeout/unavailable handling (NB6 §7)
- [ ] BigQuery fallback path when a key is missing/expired (NB3 §5)
- [ ] Training-serving skew check running on a sample (NB6 §6)

**Vector search (if used)**
- [ ] HNSW index with `EF_RUNTIME` tuned to a recall target (NB7 §6, §8)
- [ ] Hybrid filters where candidate sets can be pre-narrowed (NB7 §7)

**Operations**
- [ ] Cloud Monitoring alerts: memory usage ratio, cache hit ratio, evictions, failover
- [ ] Least-privilege IAM: separate serving / sync / admin service accounts
- [ ] Failover tested; clients verified to reconnect
- [ ] Load tested at expected peak QPS (this notebook §8)
