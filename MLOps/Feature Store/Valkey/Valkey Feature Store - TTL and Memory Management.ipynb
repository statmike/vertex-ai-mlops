# Valkey Feature Store — TTL, Eviction, and Memory Management

Feature expiration, eviction policies, and memory pressure handling. These concepts are unique to in-memory stores — Bigtable and Spanner scale storage automatically, but Valkey has a hard memory ceiling.

**What you'll learn:**
- `EXPIRE` and `TTL`: set per-key expiration times
- `PERSIST`: remove expiration from a key
- Eviction policies: `allkeys-lru`, `volatile-lru`, `allkeys-lfu`, `noeviction`
- Feature freshness gating: reject reads on expired keys
- Use cases: session features (30-min TTL), daily features (24-hour TTL), permanent features
- Memory monitoring: `INFO memory`, used_memory, fragmentation ratio
- When eviction is normal vs a problem

**Key difference from Bigtable:** Bigtable's GC policies delete old cell versions (time-based or version-count). Valkey's `EXPIRE` deletes entire keys. Bigtable scales storage automatically; Valkey has a memory ceiling — eviction is a first-class concern.

**Prerequisites:** Run [Environment](./Valkey%20Feature%20Store%20-%20Environment.ipynb) and [Fundamentals](./Valkey%20Feature%20Store%20-%20Fundamentals.ipynb) first.

## 1. EXPIRE and TTL Basics

Set a key to expire after N seconds. After expiration, the key is automatically deleted.

```python
if REDIS_AVAILABLE:
    test_key = 'ttl_demo:session:abc123'

    # Write a session feature with 10-second TTL
    decode_client.hset(test_key, mapping={
        'user_id': 'user_001',
        'page_views': '5',
        'last_click': 'product_page',
    })
    decode_client.expire(test_key, 10)  # Expire in 10 seconds

    # Check TTL
    ttl = decode_client.ttl(test_key)
    print(f'Key: {test_key}')
    print(f'TTL: {ttl} seconds remaining')
    print(f'Data: {decode_client.hgetall(test_key)}')

    # Wait and check again
    print(f'\nWaiting 5 seconds...')
    time.sleep(5)
    ttl = decode_client.ttl(test_key)
    print(f'TTL after 5s: {ttl} seconds remaining')

    print(f'\nWaiting 6 more seconds...')
    time.sleep(6)
    exists = decode_client.exists(test_key)
    print(f'Key exists: {bool(exists)} (expired and auto-deleted)')
```

## 2. TTL Patterns for Feature Stores

| Feature Type | TTL | Example |
|---|---|---|
| **Session features** | 30 minutes | Current page, cart contents, click history |
| **Real-time signals** | 5 minutes | Last-5-min click count, trending score |
| **Daily features** | 24 hours | Daily aggregates, daily model scores |
| **Batch features** | 1 hour (refreshed by sync) | Features from BigQuery sync |
| **Permanent features** | No TTL | User demographics, item catalog |

```python
if REDIS_AVAILABLE:
    # Demonstrate TTL patterns
    # All demo keys use a ttldemo: namespace so we never touch real feature data.
    patterns = [
        ('ttldemo:session:user_001', {'page_views': '12', 'last_action': 'cart_add'}, 1800),  # 30 min
        ('ttldemo:realtime:user_001', {'click_count_5m': '7', 'trending': '0.85'}, 300),  # 5 min
        ('ttldemo:daily:user_001', {'daily_spend': '142.50', 'daily_views': '23'}, 86400),  # 24 hours
        ('ttldemo:permanent:item_42', {'price': '42.5', 'category': 'electronics'}, None),  # No TTL
    ]

    print(f'TTL patterns for feature types:\n')
    for key, mapping, ttl_seconds in patterns:
        decode_client.hset(key, mapping=mapping)
        if ttl_seconds:
            decode_client.expire(key, ttl_seconds)
        actual_ttl = decode_client.ttl(key)
        ttl_label = f'{actual_ttl}s' if actual_ttl > 0 else 'permanent'
        print(f'  {key:<30s} TTL: {ttl_label}')

    # Cleanup all demo keys (the ttldemo: namespace never collides with real data)
    for key, _, ttl in patterns:
        decode_client.delete(key)
    print(f'\nCleaned up {len(patterns)} ttldemo keys (real feature data untouched).')
```

## 3. Feature Freshness Gating

Use TTL to implement freshness gating: only serve features if they're within their freshness window.

```python
if REDIS_AVAILABLE:
    def read_features_with_freshness(key, max_age_seconds=3600):
        """Read features, reject if expired or too stale."""
        # Check if key exists
        if not decode_client.exists(key):
            return None, 'missing'

        # Check TTL (if set)
        ttl = decode_client.ttl(key)
        if ttl == -2:  # Key doesn't exist
            return None, 'expired'

        # Read features
        data = decode_client.hgetall(key)
        return data, 'fresh'

    # Test
    data, status = read_features_with_freshness('features:A:000001')
    print(f'features:A:000001: status={status}, fields={len(data) if data else 0}')

    data, status = read_features_with_freshness('features:NONEXISTENT:999999')
    print(f'features:NONEXISTENT:999999: status={status}')
```

## 4. Memory Monitoring

Monitor memory usage, fragmentation, and eviction metrics.

```python
if REDIS_AVAILABLE:
    mem = decode_client.info('memory')

    print(f'Memory stats:')
    print(f'  used_memory:          {mem["used_memory_human"]}')
    print(f'  used_memory_peak:     {mem["used_memory_peak_human"]}')
    print(f'  used_memory_dataset:  {mem.get("used_memory_dataset_human", "N/A")}')
    print(f'  mem_fragmentation:    {mem["mem_fragmentation_ratio"]:.2f}')
    print(f'  maxmemory:            {mem.get("maxmemory_human", "N/A")}')
    print(f'  maxmemory_policy:     {mem.get("maxmemory_policy", "N/A")}')

    stats = decode_client.info('stats')
    print(f'\nKey stats:')
    print(f'  total_keys:           {decode_client.dbsize():,}')
    print(f'  evicted_keys:         {stats.get("evicted_keys", 0):,}')
    print(f'  expired_keys:         {stats.get("expired_keys", 0):,}')

    # Fragmentation interpretation
    frag = mem['mem_fragmentation_ratio']
    if frag < 1.0:
        print(f'\n\u26a0 Fragmentation < 1.0: Redis is using swap (slow)')
    elif frag < 1.5:
        print(f'\n\u2714 Fragmentation {frag:.2f}: healthy')
    else:
        print(f'\n\u26a0 Fragmentation {frag:.2f}: high (consider restart to compact)')
```

## 5. Eviction Policies

When Valkey reaches `maxmemory`, the eviction policy determines which keys are removed.

| Policy | Description | Best For |
|---|---|---|
| `noeviction` | Reject writes when full | When data loss is unacceptable |
| `allkeys-lru` | Evict least recently used keys | General purpose |
| `volatile-lru` | Evict LRU keys with TTL set | Mix of permanent + ephemeral features |
| `allkeys-lfu` | Evict least frequently used keys | Access-frequency-based |
| `volatile-ttl` | Evict keys with shortest remaining TTL | Time-sensitive features |

**For feature stores:** `volatile-lru` is typically best — it evicts ephemeral features (with TTL) before permanent ones (without TTL). This keeps your core features in memory while allowing session/real-time features to be reclaimed under pressure.

```python
if REDIS_AVAILABLE:
    # Managed Memorystore disables the CONFIG command, so we read settings from
    # INFO memory instead of CONFIG GET. INFO always works; CONFIG GET/SET does not.
    mem = decode_client.info('memory')
    maxmem = mem.get('maxmemory_human', '(instance limit)')
    policy = mem.get('maxmemory_policy', '(set at instance level)')

    print(f'Current eviction config (read via INFO, not CONFIG):')
    print(f'  maxmemory:        {maxmem}')
    print(f'  maxmemory-policy: {policy}')

    print(f'\nRecommendation for feature stores:')
    print(f'  volatile-lru: evicts TTL keys first, preserves permanent features.')
    print(f'\nManaged Memorystore: CONFIG SET is disabled. Set the policy at the')
    print(f'  instance level instead:')
    print(f'    Terraform: redis_configs = {{ "maxmemory-policy" = "volatile-lru" }}')
    print(f'    gcloud:    gcloud redis instances update INSTANCE \\')
    print(f'                 --update-redis-config maxmemory-policy=volatile-lru')
    print(f'\n  On self-hosted Redis/Valkey (e.g., local Docker), CONFIG SET works:')
    print(f'    CONFIG SET maxmemory-policy volatile-lru')
```

## 6. Persistence — What Survives a Restart

Bigtable and Spanner are always durable (disk-backed). Valkey is **in-memory** — durability is a configuration choice. For a feature store this is less critical than it sounds, because **BigQuery is the durable source of truth** — but understanding persistence determines your cold-start re-sync cost.

| Mechanism | What it does | Trade-off |
|---|---|---|
| **None** | Pure cache | Fastest; lose everything on restart -> full re-sync from BigQuery |
| **RDB snapshot** | Periodic point-in-time dump | Fast restart; lose writes since last snapshot |
| **AOF** | Append-only log of every write | Near-zero loss; larger files, slower restart |
| **HA replication** | Standard tier replicates to a replica | Failover without data loss (Memorystore's primary durability) |

```python
if REDIS_AVAILABLE:
    # Persistence config is reported in INFO persistence (CONFIG is disabled on Memorystore)
    persist = decode_client.info('persistence')
    print('Persistence state (from INFO persistence):')
    print(f'  rdb_last_save_time:        {persist.get("rdb_last_save_time")}')
    print(f'  rdb_changes_since_save:    {persist.get("rdb_changes_since_save")}')
    print(f'  rdb_last_bgsave_status:    {persist.get("rdb_last_bgsave_status")}')
    print(f'  aof_enabled:               {persist.get("aof_enabled")}')
    print(f'  loading:                   {persist.get("loading")}')

    print('\nWhat survives a restart:')
    print('  RDB snapshot: point-in-time dump, written periodically (or on BGSAVE).')
    print('               Fast restart, but you can lose writes since the last snapshot.')
    print('  AOF (append-only file): logs every write; replayed on restart.')
    print('               Near-zero data loss, larger files, slower restart.')
    print('\nMemorystore manages persistence for you:')
    print('  - Standard HA tier replicates to a replica for failover (this is the')
    print('    primary durability mechanism, not RDB/AOF).')
    print('  - RDB snapshots can be enabled for point-in-time recovery via the')
    print('    instance config (persistence-mode = RDB, with a snapshot period).')
    print('\nFor a feature store this matters less than you might think:')
    print('  BigQuery is the durable source of truth. If Valkey loses data, re-run')
    print('  the sync (NB3). Persistence mainly reduces cold-start re-sync time.')
```

## 7. Active vs Lazy Expiration *(operational gotcha)*

Setting a TTL doesn't delete the key at that instant. Redis expires keys two ways:
- **Lazy** — when a key is *accessed* past its TTL, it's deleted then and returns nil
- **Active** — a background cycle *samples* keys with TTLs and evicts the expired ones

The consequence: **expired-but-not-yet-collected keys still consume memory**. Under high TTL churn (5-second quotes, short-lived session features), `used_memory` lags behind logical expiry. This is a real sizing consideration unique to in-memory stores.

```python
if REDIS_AVAILABLE:
    # Demonstrate that an expired key still counts in stats until collected.
    # Redis expires keys two ways:
    #   1. Lazy:   on access, if a key is past its TTL it's deleted and returns nil
    #   2. Active: a background cycle samples keys with TTLs and evicts expired ones
    import time

    # Write 100 keys with a 1-second TTL
    pipe = decode_client.pipeline(transaction=False)
    for i in range(100):
        k = f'expdemo:{i:03d}'
        pipe.set(k, 'x', ex=1)
    pipe.execute()

    before = decode_client.info('stats').get('expired_keys', 0)
    print(f'Wrote 100 keys with 1s TTL. expired_keys so far: {before:,}')

    time.sleep(2)  # let them all pass their TTL

    # At this moment many are expired but may not be collected yet.
    # A lazy access forces collection of the touched key:
    sample = decode_client.get('expdemo:000')  # past TTL -> returns None, deletes lazily
    print(f'Lazy access of expired key returns: {sample}')

    # Active expiration runs in the background; give it a moment, then check stats
    time.sleep(1)
    after = decode_client.info('stats').get('expired_keys', 0)
    print(f'expired_keys after TTL elapsed: {after:,} (delta {after-before:,})')

    print('\nWhy this matters for a feature store:')
    print('  Expired-but-not-yet-collected keys still occupy memory. Under high TTL')
    print('  churn (e.g., 5s quotes), monitor used_memory vs key count — active')
    print('  expiration is sampling-based, so memory lags behind logical expiry.')

    # cleanup any stragglers
    for i in range(100):
        decode_client.delete(f'expdemo:{i:03d}')
```

## 8. Hash-Field TTL (HEXPIRE) *(Redis 7.4+)*

Whole-key `EXPIRE` expires an entire entity hash. But often you want to expire **one feature** within an entity — e.g., a real-time score valid for only 30 seconds — while keeping the stable features. Redis 7.4+ adds `HEXPIRE`/`HTTL`/`HPERSIST` for per-field expiration.

This maps cleanly to feature stores: mix permanent features (demographics) and short-lived features (live signals) in the **same hash**, each with its own lifetime.

```python
if REDIS_AVAILABLE:
    # Redis 7.4+ supports per-field TTL on hashes (HEXPIRE / HTTL / HPERSIST).
    # This lets you expire ONE feature in an entity hash while keeping the rest —
    # impossible with whole-key EXPIRE.
    key = 'hexpire:demo'
    decode_client.delete(key)
    decode_client.hset(key, mapping={
        'stable_feature': 'always_here',
        'realtime_score': '0.92',     # short-lived
        'session_flag': 'active',     # short-lived
    })

    try:
        # Expire only the two short-lived fields in 2 seconds
        res = decode_client.hexpire(key, 2, 'realtime_score', 'session_flag')
        print(f'HEXPIRE result (1=set): {res}')

        ttls = decode_client.httl(key, 'stable_feature', 'realtime_score', 'session_flag')
        print(f'Per-field TTL: stable={ttls[0]}, realtime={ttls[1]}, session={ttls[2]}')
        print('  (-1 = no TTL on that field, positive = seconds remaining)')

        import time
        time.sleep(3)
        remaining = decode_client.hgetall(key)
        print(f'\nAfter 3s, fields remaining: {list(remaining.keys())}')
        print('  -> only stable_feature survived; the two timed fields expired individually.')
        HEXPIRE_OK = True
    except redis.ResponseError as e:
        HEXPIRE_OK = False
        print(f'HEXPIRE not supported on this server version: {e}')
        print('  Hash-field TTL requires Redis 7.4+. On older servers, model short-lived')
        print('  features as separate keys (features:realtime:{id}) with whole-key EXPIRE.')

    decode_client.delete(key)
```

## 9. No Native History — Use BigQuery for Time-Travel

Bigtable stores feature **history** via cell versioning (multiple timestamped values per cell), enabling point-in-time joins for training. Valkey has **no native versioning** — a hash field holds exactly one current value.

This is by design: Valkey is the **current-state serving layer**, not the historical store. The division of labor:

| Need | Where |
|---|---|
| Current feature value (serving) | **Valkey** (sub-ms) |
| Feature history / point-in-time joins (training) | **BigQuery** (the offline store — already has full history) |
| Time-travel queries | **BigQuery** `FOR SYSTEM_TIME AS OF` |

If you need a short window of recent values in Valkey (e.g., last-5 prices), model it explicitly with a **sorted set** (timestamp as score) or a **Redis Stream** — covered in NB5. But for full history, read from BigQuery; don't try to make Valkey a historical store.

---
## 10. Comparison to Bigtable Data Lifecycle

| Aspect | Valkey | Bigtable |
|---|---|---|
| **Expiration unit** | Entire key (EXPIRE) | Cell versions (GC policy) |
| **Granularity** | Per-key TTL | Per-column-family GC rules |
| **Storage scaling** | Hard memory ceiling → eviction | Automatic disk scaling |
| **Memory pressure** | Eviction (data loss) | N/A (disk-based) |
| **Monitoring** | INFO memory, evicted_keys | Cloud Monitoring, disk utilization |
| **Best practice** | Right-size memory, use TTL + volatile-lru | Set GC policies, monitor storage |

**Key insight:** In Valkey, TTL is a **design tool** — you use it intentionally to manage memory and express feature freshness requirements. In Bigtable, GC policies are cleanup mechanisms for cell version history.
