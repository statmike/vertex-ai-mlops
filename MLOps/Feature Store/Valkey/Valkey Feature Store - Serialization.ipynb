# Valkey Feature Store — Serialization and Data Types

How to encode BigQuery's 19 data types into Redis strings and bytes. Unlike Spanner (which has native typed columns) and Bigtable (which has 5 serialization strategies), Valkey stores everything as strings or raw bytes — but the encoding choices still matter for storage size and read latency.

**What you'll learn:**
- Simple types: INT64, FLOAT64, STRING, BOOL → string representation
- Complex types: ARRAY, STRUCT, JSON → JSON string encoding
- Embeddings: ARRAY<FLOAT64> → `struct.pack` float32 bytes (required for FT.SEARCH)
- Timestamps/dates → ISO 8601 strings or Unix epoch integers
- NULL handling: field absence vs sentinel values
- Storage size comparison across encoding strategies
- Read latency comparison: parsing overhead for each encoding

**Key difference from Bigtable:** The Bigtable series compares 5 serialization methods (native, JSON, concatenation, protobuf, hybrid). Valkey is simpler: strings for scalars, JSON for complex types, packed bytes for vectors. But encoding decisions still affect read latency and memory usage.

**Prerequisites:** Run [Environment](./Valkey%20Feature%20Store%20-%20Environment.ipynb) and [Fundamentals](./Valkey%20Feature%20Store%20-%20Fundamentals.ipynb) first.

## 2. Encoding Functions

```python
def encode_for_redis(value, bq_type=None):
    """Encode a BigQuery value for Redis hash storage."""
    # Handle None and collections before pd.isna (which returns an array for lists)
    if value is None:
        return None

    if isinstance(value, (list, dict, np.ndarray)):
        # numpy arrays come back from BigQuery ARRAY columns
        if isinstance(value, np.ndarray):
            value = value.tolist()
        # default=str handles nested dates/timestamps inside structs/arrays/ranges
        return json.dumps(value, default=str)

    # Scalar null check (safe now that collections are handled)
    if pd.isna(value):
        return None

    if isinstance(value, (bool, np.bool_)):
        return str(bool(value))

    if isinstance(value, (int, np.integer)):
        return str(int(value))

    if isinstance(value, (float, np.floating)):
        return str(float(value))

    if isinstance(value, str):
        return value

    # Dates, datetimes, timestamps
    if hasattr(value, 'isoformat'):
        return value.isoformat()

    return str(value)


def decode_from_redis(value, target_type):
    """Decode a Redis string value to a Python type."""
    if value is None:
        return None

    if target_type == 'int':
        return int(value)
    if target_type == 'float':
        return float(value)
    if target_type == 'bool':
        return value == 'True'
    if target_type == 'json':
        return json.loads(value)
    return value  # string as-is


# Test encoding
test_values = [
    (42, 'int'),
    (3.14159, 'float'),
    ('hello', 'str'),
    (True, 'bool'),
    ([1, 2, 3], 'list'),
    ({'score': 42.0, 'label': 'positive'}, 'dict'),
]

print('Encoding examples:')
for val, label in test_values:
    encoded = encode_for_redis(val)
    print(f'  {label:<6s} {str(val):<40s} → "{encoded}" ({len(encoded)} bytes)')
```

## 3. Embedding Encoding — struct.pack

For vector search with FT.SEARCH, embeddings must be stored as raw float32 bytes (not JSON strings). The `struct.pack` format packs each float as 4 bytes.

```python
def encode_embedding(embedding, dtype='float32'):
    """Encode a list of floats to packed bytes for FT.SEARCH."""
    arr = np.array(embedding, dtype=np.float32)
    return arr.tobytes()


def decode_embedding(raw_bytes, dim=None):
    """Decode packed bytes back to numpy array."""
    return np.frombuffer(raw_bytes, dtype=np.float32)


# Test
test_embedding = np.random.randn(64).tolist()
encoded = encode_embedding(test_embedding)
decoded = decode_embedding(encoded)

print(f'Embedding encoding:')
print(f'  Input:  {len(test_embedding)} floats (Python list)')
print(f'  Packed: {len(encoded)} bytes ({len(encoded) / len(test_embedding):.0f} bytes/float)')
print(f'  Match:  {np.allclose(test_embedding, decoded, atol=1e-6)}')

# Compare to JSON encoding
json_encoded = json.dumps(test_embedding)
print(f'\nSize comparison for 64-dim embedding:')
print(f'  struct.pack (float32): {len(encoded):>6} bytes')
print(f'  JSON string:          {len(json_encoded):>6} bytes')
print(f'  Ratio: JSON is {len(json_encoded) / len(encoded):.1f}x larger')
```

## 4. Five Serialization Methods

The [Bigtable Feature Store serialization notebook](../Bigtable/Bigtable%20Feature%20Store%20-%20Serialization.ipynb) compares five methods for encoding features into a byte-oriented store. All five apply to Redis/Valkey hashes too — and the trade-offs are sharpened by Redis's access model:

| Method | Storage in Redis | Read pattern |
|---|---|---|
| **1. Native hash fields** | One hash field per feature | `HGET` any single field without reading the rest |
| **2. JSON blob** | Entire feature set as one JSON string field | Must read + parse the whole blob |
| **3. Concatenation** | Delimited string in one field | Must read + split the whole string |
| **4. Protocol Buffers** | Serialized protobuf bytes in one field | Must read + deserialize the whole message |
| **5. Hybrid** | Hot features native + cold features in a protobuf blob | `HGET` hot features directly; deserialize blob only when needed |

**The key Redis-specific difference:** native hash fields let you fetch a single feature with `HGET key field` — Redis returns just that value. Blob methods (JSON, concat, protobuf) require reading the entire serialized payload even if you only need one feature. This is the central serving trade-off.

We'll encode the same 100 entities with each method, then compare storage size and read latency.

```python
# Load a common set of entities to encode with every method
N_ENTITIES = 100

source_df = bq.query(f'''
    SELECT *
    FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
    WHERE entity_group = 'A' AND entity_id <= '000100'
    ORDER BY entity_id
''').to_dataframe()

# Feature columns = everything except the key columns
KEY_COLS = ['entity_group', 'entity_id']
FEATURE_COLS = [c for c in source_df.columns if c not in KEY_COLS]

print(f'Loaded {len(source_df)} entities with {len(FEATURE_COLS)} feature columns')
print(f'Each method will encode these same entities under a different key prefix.')
```

---
## 5. Method 1 — Native Hash Fields

Store each feature as its own hash field. This is the default pattern (used in NB1 Fundamentals). Complex types (arrays, structs, JSON) are encoded as strings via `encode_for_redis`.

**Pro:** `HGET key field` fetches a single feature — no need to read the whole entity.
**Con:** Many small fields add per-field memory overhead in Redis.

```python
def write_native(client, source_df):
    pipe = client.pipeline(transaction=False)
    for _, row in source_df.iterrows():
        key = f'serial:native:{row["entity_group"]}:{row["entity_id"]}'
        mapping = {}
        for col in FEATURE_COLS:
            enc = encode_for_redis(row[col])
            if enc is not None:
                mapping[col] = enc
        if mapping:
            pipe.hset(key, mapping=mapping)
    pipe.execute()

if REDIS_AVAILABLE:
    write_native(decode_client, source_df)

    # Read back a single feature without reading the rest
    sample_key = 'serial:native:A:000001'
    one_field = decode_client.hget(sample_key, 'feature_float_1')
    field_count = decode_client.hlen(sample_key)
    print(f'Method 1 — Native hash fields:')
    print(f'  Wrote {len(source_df)} entities')
    print(f'  Sample {sample_key}: {field_count} fields')
    print(f'  HGET single field (feature_float_1): {one_field}')
    print(f'  → Single-field read touches only that field.')
```

---
## 6. Method 2 — JSON Blob

Serialize the entire feature set into one JSON string, stored in a single hash field (or as a plain string key). Self-describing and trivial to implement.

**Pro:** One field, human-readable, schema travels with the data.
**Con:** Must read and parse the entire blob to access any single feature; JSON is verbose (numbers as text, field names repeated).

```python
def row_to_dict(row):
    d = {}
    for col in FEATURE_COLS:
        enc = encode_for_redis(row[col])
        if enc is not None:
            d[col] = enc
    return d

def write_json_blob(client, source_df):
    pipe = client.pipeline(transaction=False)
    for _, row in source_df.iterrows():
        key = f'serial:json:{row["entity_group"]}:{row["entity_id"]}'
        blob = json.dumps(row_to_dict(row))
        pipe.set(key, blob)
    pipe.execute()

if REDIS_AVAILABLE:
    write_json_blob(decode_client, source_df)

    sample_key = 'serial:json:A:000001'
    blob = decode_client.get(sample_key)
    parsed = json.loads(blob)
    print(f'Method 2 — JSON blob:')
    print(f'  Wrote {len(source_df)} entities as JSON strings')
    print(f'  Sample {sample_key}: {len(blob)} bytes, {len(parsed)} fields after parse')
    print(f'  feature_float_1 = {parsed.get("feature_float_1")}')
    print(f'  → Reading one feature requires parsing the whole blob.')
```

---
## 7. Method 3 — Pipe-Delimited Concatenation

Concatenate all feature values into a single delimited string, with a companion schema defining field order. Compact but fragile.

**Pro:** Smallest text representation, single field.
**Con:** The **delimiter trap** — if any value contains the delimiter, parsing breaks. Requires a fixed field-order schema. No self-description.

```python
DELIMITER = '|'

# Schema = the ordered list of fields (must be stored alongside the data)
CONCAT_SCHEMA = FEATURE_COLS

def write_concat(client, source_df):
    # Store the field-order schema once
    client.set('serial:concat:__schema__', DELIMITER.join(CONCAT_SCHEMA))
    pipe = client.pipeline(transaction=False)
    for _, row in source_df.iterrows():
        key = f'serial:concat:{row["entity_group"]}:{row["entity_id"]}'
        parts = []
        for col in CONCAT_SCHEMA:
            enc = encode_for_redis(row[col])
            # Empty string for NULL; strip delimiter to avoid corruption (the trap!)
            parts.append('' if enc is None else enc.replace(DELIMITER, ' '))
        pipe.set(key, DELIMITER.join(parts))
    pipe.execute()

if REDIS_AVAILABLE:
    write_concat(decode_client, source_df)

    schema = decode_client.get('serial:concat:__schema__').split(DELIMITER)
    sample_key = 'serial:concat:A:000001'
    raw = decode_client.get(sample_key)
    values = raw.split(DELIMITER)
    parsed = dict(zip(schema, values))
    print(f'Method 3 — Pipe-delimited concatenation:')
    print(f'  Wrote {len(source_df)} entities as delimited strings')
    print(f'  Sample {sample_key}: {len(raw)} bytes, {len(values)} values')
    print(f'  feature_float_1 = {parsed.get("feature_float_1")}')
    print(f'  ⚠ The delimiter trap: any value containing "{DELIMITER}" would corrupt parsing.')
    print(f'    We strip the delimiter on write — lossy but safe.')
```

---
## 8. Method 4 — Protocol Buffers

Encode features as a binary protobuf message. We build the schema dynamically from the BigQuery column types (no `.proto` file needed) and store the serialized `FileDescriptorSet` so readers can decode without a compiled schema — exactly the self-describing pattern from the Bigtable notebook.

**Pro:** Compact binary, strongly typed, schema can be embedded for self-description.
**Con:** Must deserialize the whole message to read any field; requires the protobuf library; encoding is more complex.

```python
from google.protobuf import descriptor_pb2
from google.protobuf import descriptor_pool
from google.protobuf import message_factory

# Map BigQuery types to protobuf field types; complex types serialize as STRING
BQ_TO_PROTO = {
    'INT64': descriptor_pb2.FieldDescriptorProto.TYPE_INT64,
    'FLOAT64': descriptor_pb2.FieldDescriptorProto.TYPE_DOUBLE,
    'STRING': descriptor_pb2.FieldDescriptorProto.TYPE_STRING,
    'BOOL': descriptor_pb2.FieldDescriptorProto.TYPE_BOOL,
    'BYTES': descriptor_pb2.FieldDescriptorProto.TYPE_BYTES,
}

# Look up each feature column's BigQuery type
col_types = bq.query(f'''
    SELECT column_name, data_type
    FROM `{BQ_PROJECT}.{BQ_DATASET}.INFORMATION_SCHEMA.COLUMNS`
    WHERE table_name = 'dense_features'
''').to_dataframe()
type_map = dict(zip(col_types['column_name'], col_types['data_type']))

# Build a FileDescriptorProto for our feature row
file_proto = descriptor_pb2.FileDescriptorProto(
    name='valkey_features.proto', package='valkeyfs', syntax='proto3',
    message_type=[descriptor_pb2.DescriptorProto(
        name='FeatureRow',
        field=[
            descriptor_pb2.FieldDescriptorProto(
                name=col, number=i + 1,
                type=BQ_TO_PROTO.get(type_map.get(col, 'STRING').split('(')[0].split('<')[0].strip(),
                                     descriptor_pb2.FieldDescriptorProto.TYPE_STRING),
                label=descriptor_pb2.FieldDescriptorProto.LABEL_OPTIONAL,
            )
            for i, col in enumerate(FEATURE_COLS)
        ],
    )],
)

# Serialize the FileDescriptorSet — this is what makes it self-describing
descriptor_set_bytes = descriptor_pb2.FileDescriptorSet(file=[file_proto]).SerializeToString()

# Build the message class
pool = descriptor_pool.DescriptorPool()
pool.Add(file_proto)
FeatureRow = message_factory.GetMessageClass(pool.FindMessageTypeByName('valkeyfs.FeatureRow'))

print(f'Built protobuf schema: FeatureRow with {len(FeatureRow.DESCRIPTOR.fields)} fields')
print(f'  FileDescriptorSet: {len(descriptor_set_bytes)} bytes (stored for self-description)')
```

```python
def proto_field_type(col):
    return BQ_TO_PROTO.get(type_map.get(col, 'STRING').split('(')[0].split('<')[0].strip(),
                           descriptor_pb2.FieldDescriptorProto.TYPE_STRING)

def write_protobuf(client_bytes, source_df):
    # Store the descriptor set once so readers can rebuild the schema
    client_bytes.set(b'serial:proto:__descriptor__', descriptor_set_bytes)
    pipe = client_bytes.pipeline(transaction=False)
    for _, row in source_df.iterrows():
        msg = FeatureRow()
        for col in FEATURE_COLS:
            val = row[col]
            enc = encode_for_redis(val)
            if enc is None:
                continue
            ftype = proto_field_type(col)
            if ftype == descriptor_pb2.FieldDescriptorProto.TYPE_INT64:
                setattr(msg, col, int(float(enc)))
            elif ftype == descriptor_pb2.FieldDescriptorProto.TYPE_DOUBLE:
                setattr(msg, col, float(enc))
            elif ftype == descriptor_pb2.FieldDescriptorProto.TYPE_BOOL:
                setattr(msg, col, enc == 'True')
            elif ftype == descriptor_pb2.FieldDescriptorProto.TYPE_BYTES:
                setattr(msg, col, enc.encode() if isinstance(enc, str) else enc)
            else:
                setattr(msg, col, str(enc))
        key = f'serial:proto:{row["entity_group"]}:{row["entity_id"]}'.encode()
        pipe.set(key, msg.SerializeToString())
    pipe.execute()

if REDIS_AVAILABLE:
    write_protobuf(bytes_client, source_df)

    # Reader side: rebuild schema from stored descriptor, then deserialize
    stored_desc = bytes_client.get(b'serial:proto:__descriptor__')
    fds = descriptor_pb2.FileDescriptorSet.FromString(stored_desc)
    reader_pool = descriptor_pool.DescriptorPool()
    reader_pool.Add(fds.file[0])
    ReaderRow = message_factory.GetMessageClass(reader_pool.FindMessageTypeByName('valkeyfs.FeatureRow'))

    raw = bytes_client.get(b'serial:proto:A:000001')
    msg = ReaderRow.FromString(raw)
    print(f'Method 4 — Protocol Buffers:')
    print(f'  Wrote {len(source_df)} entities as protobuf bytes')
    print(f'  Sample serial:proto:A:000001: {len(raw)} bytes')
    print(f'  feature_float_1 = {msg.feature_float_1:.4f}')
    print(f'  → Reader rebuilt the schema from the stored FileDescriptorSet — no .proto file needed.')
```

## 9. Method 5 — Hybrid (Hot Native + Cold Blob)

Store the most-frequently-accessed features as native hash fields (fast single-field `HGET`), and pack the long tail of rarely-accessed features into a protobuf blob in one field. This optimizes for the common access pattern while keeping storage compact.

**Pro:** Hot features get fast per-field reads; cold features stay compact.
**Con:** Two encodings to maintain; the split must match the real access pattern.

```python
# Hot = the 10 most-accessed features (here: the typed scalars from NB1)
HOT_FEATURES = [
    'feature_int_1', 'feature_int_2', 'feature_int_3',
    'feature_float_1', 'feature_float_2', 'feature_float_5',
    'feature_bool_1', 'feature_bool_2', 'feature_bool_3', 'label',
]
HOT_FEATURES = [c for c in HOT_FEATURES if c in FEATURE_COLS]
COLD_FEATURES = [c for c in FEATURE_COLS if c not in HOT_FEATURES]

# Build a cold-only protobuf schema
cold_file_proto = descriptor_pb2.FileDescriptorProto(
    name='valkey_cold.proto', package='valkeyfs', syntax='proto3',
    message_type=[descriptor_pb2.DescriptorProto(
        name='ColdRow',
        field=[descriptor_pb2.FieldDescriptorProto(
            name=col, number=i + 1, type=proto_field_type(col),
            label=descriptor_pb2.FieldDescriptorProto.LABEL_OPTIONAL,
        ) for i, col in enumerate(COLD_FEATURES)],
    )],
)
cold_pool = descriptor_pool.DescriptorPool()
cold_pool.Add(cold_file_proto)
ColdRow = message_factory.GetMessageClass(cold_pool.FindMessageTypeByName('valkeyfs.ColdRow'))

def write_hybrid(client_bytes, source_df):
    pipe = client_bytes.pipeline(transaction=False)
    for _, row in source_df.iterrows():
        key = f'serial:hybrid:{row["entity_group"]}:{row["entity_id"]}'.encode()
        mapping = {}
        # Hot features as native fields
        for col in HOT_FEATURES:
            enc = encode_for_redis(row[col])
            if enc is not None:
                mapping[col.encode()] = enc.encode()
        # Cold features as a protobuf blob in one field
        cold = ColdRow()
        for col in COLD_FEATURES:
            enc = encode_for_redis(row[col])
            if enc is None:
                continue
            ftype = proto_field_type(col)
            if ftype == descriptor_pb2.FieldDescriptorProto.TYPE_INT64:
                setattr(cold, col, int(float(enc)))
            elif ftype == descriptor_pb2.FieldDescriptorProto.TYPE_DOUBLE:
                setattr(cold, col, float(enc))
            elif ftype == descriptor_pb2.FieldDescriptorProto.TYPE_BOOL:
                setattr(cold, col, enc == 'True')
            elif ftype == descriptor_pb2.FieldDescriptorProto.TYPE_BYTES:
                setattr(cold, col, enc.encode() if isinstance(enc, str) else enc)
            else:
                setattr(cold, col, str(enc))
        mapping[b'__cold__'] = cold.SerializeToString()
        pipe.hset(key, mapping=mapping)
    pipe.execute()

if REDIS_AVAILABLE:
    write_hybrid(bytes_client, source_df)

    key = b'serial:hybrid:A:000001'
    # Hot read: single field, no deserialization
    hot_val = bytes_client.hget(key, b'feature_float_1')
    # Cold read: fetch + deserialize the blob
    cold_blob = bytes_client.hget(key, b'__cold__')
    cold_msg = ColdRow.FromString(cold_blob)
    print(f'Method 5 — Hybrid:')
    print(f'  Hot features ({len(HOT_FEATURES)}): native fields')
    print(f'  Cold features ({len(COLD_FEATURES)}): protobuf blob')
    print(f'  HGET hot feature_float_1: {hot_val.decode()} (no deserialization)')
    print(f'  Cold blob: {len(cold_blob)} bytes → deserialize only when a cold feature is needed')
```

## 12. Recommendation Matrix

Based on the measured results above (222 feature columns, 100 entities):

| Method | Storage | Single-field read | All-field read | Self-describing | Best for |
|---|---|---|---|---|---|
| **Native fields** | Highest (per-field overhead) | **Fast** (single HGET) | **Slowest** (HGETALL returns all 222 fields) | Field names = schema | Partial reads — fetch a few features per request |
| **JSON blob** | Large | Slowest (parse whole blob) | Medium | Yes | Debuggability, schema flux, small feature sets |
| **Concatenation** | **Smallest** | Slow (split whole string) | Medium | No (needs schema) | Extreme storage pressure, fixed schema |
| **Protocol Buffers** | Small | Slow (deserialize whole msg) | Medium | Yes (descriptor) | Large feature sets read whole; cross-language |
| **Hybrid** | High | **Fastest** for hot features | **Fastest** overall | Partial | Skewed access: few hot features, long cold tail |

**The central trade-off this notebook demonstrates:**
- **Native fields win when you read a few features per request** — `HGET key field` returns just that value. But `HGETALL` on a wide hash is the *slowest* all-field read, because Redis ships every field name + value over the wire.
- **Blob methods (JSON, concat, protobuf) win when you always read the whole entity** — one value over the wire, decode locally. But they can't fetch a single feature without reading everything.
- **Hybrid gets the best of both** for skewed access patterns: hot features as native fields (fast `HGET`), the cold tail as one protobuf blob (compact, single fetch).

**For Valkey feature serving, choose by access pattern, not by default:**
- Serving requests that read a handful of features → **native fields** or **hybrid**
- Serving requests that always pull the full feature vector → **protobuf** or **hybrid**
- Storage-constrained, fixed schema, whole-entity reads → **concatenation**

**Compared to Bigtable:** the same five methods apply, but Redis's `HGET`-single-field capability is the key difference — it makes native fields genuinely attractive for partial reads, whereas in Bigtable reading any cell already pulls the column family. The blob trade-off (read-everything-to-read-anything) is identical in both stores.

## 13. NULL Handling

Redis hashes have no NULL concept — a field either exists or it doesn't. Two strategies: **field absence** (skip NULL fields, check with `HGET → None`) or **sentinel values** (store a marker). Field absence is the natural Redis pattern and saves memory.

```python
if REDIS_AVAILABLE:
    # Demonstrate field-absence NULL handling
    test_key = 'test:null_handling'

    # Write with NULLs skipped
    features = {
        'feature_float_1': '42.5',
        'feature_float_3': None,  # NULL — skip this field
        'feature_string_1': 'cat_5',
        'feature_string_2': None,  # NULL — skip
    }

    mapping = {k: v for k, v in features.items() if v is not None}
    decode_client.hset(test_key, mapping=mapping)

    # Read back
    result = decode_client.hgetall(test_key)
    print(f'Stored fields (NULLs skipped):')
    for k, v in sorted(result.items()):
        print(f'  {k}: {v}')

    # Check for NULL fields
    for field in ['feature_float_1', 'feature_float_3', 'feature_string_1', 'feature_string_2']:
        val = decode_client.hget(test_key, field)
        is_null = val is None
        print(f'  {field}: {val} (NULL={is_null})')

    decode_client.delete(test_key)

    print(f'\nRecommendation: use field absence for NULLs — saves memory and is the natural Redis pattern.')
```
