# Valkey Feature Store — Synchronization

Keeping Valkey in sync with BigQuery. Unlike Bigtable (native `EXPORT DATA`) and Spanner (native continuous query `INSERT INTO`), Valkey has **no native BigQuery export** — you build the sync bridge yourself with Pub/Sub. This notebook covers the full spectrum of sync patterns, from one-time backfill to continuous streaming, plus the Valkey-specific concerns: subscriber deployment, TTL-on-sync, dual-write, and memory pressure.

**What you'll learn:**
- **Section 1** — One-time backfill: BigQuery → pipelined HSET
- **Section 2** — Scheduled sync: Cloud Scheduler → Cloud Run → pipeline
- **Section 3** — Continuous sync: BigQuery Continuous Query → Pub/Sub → subscriber
- **Section 4** — Subscriber deployment: pull vs push, Cloud Run *(Valkey-specific)*
- **Section 5** — TTL-on-sync: automatic staleness bounds *(Valkey-specific)*
- **Section 6** — Dual-write: app writes to Valkey + BigQuery
- **Section 7** — The propagation race: compare all patterns
- **Section 8** — Memory pressure during bulk sync *(Valkey-specific)*
- **Section 9** — Reservation management & batch-read analogs
- **Section 10** — Error handling, idempotency, dead-letter topics

**Key difference from Bigtable:** Bigtable's `EXPORT DATA` is one SQL command. Valkey's sync is application code — more work, but the Pub/Sub bridge pattern is reusable for any target without native BigQuery export.

**Prerequisites:** Run [Environment](./Valkey%20Feature%20Store%20-%20Environment.ipynb) first.

## Section 1: One-Time Backfill — BigQuery to Valkey

The simplest sync: read the full table from BigQuery and write to Valkey with pipelined HSET. This is the equivalent of Bigtable's `EXPORT DATA` one-time export, but implemented in application code.

Use this for initial load, full refreshes, or rebuilding Valkey after a flush. NB1 (Fundamentals) used this pattern to load all 130K entities; here we frame it as the "one-time" sync strategy and measure it as a baseline.

````python
# One-time backfill of a single entity group as a demonstration
# (NB1 already loaded all 130K; this shows the pattern + measures throughput)

BACKFILL_COLUMNS = [
    'entity_group', 'entity_id',
    'feature_int_1', 'feature_float_1', 'feature_float_2',
    'feature_string_1', 'feature_bool_1', 'label',
]

if REDIS_AVAILABLE:
    cols_sql = ', '.join(BACKFILL_COLUMNS)
    df = bq.query(f'''
        SELECT {cols_sql}
        FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
        WHERE entity_group = 'A'
        ORDER BY entity_id
    ''').to_dataframe()

    start = time.perf_counter()
    pipe = decode_client.pipeline(transaction=False)
    for i, (_, row) in enumerate(df.iterrows()):
        key = f'features:{row["entity_group"]}:{row["entity_id"]}'
        mapping = {c: str(row[c]) for c in BACKFILL_COLUMNS[2:] if not pd.isna(row[c])}
        pipe.hset(key, mapping=mapping)
        if (i + 1) % 500 == 0:
            pipe.execute()
            pipe = decode_client.pipeline(transaction=False)
    pipe.execute()
    elapsed = time.perf_counter() - start

    print(f'One-time backfill (group A):')
    print(f'  Entities:   {len(df):,}')
    print(f'  Time:       {elapsed:.2f}s')
    print(f'  Throughput: {len(df)/elapsed:,.0f} entities/s (pipelined HSET)')
    print(f'\nEquivalent to Bigtable EXPORT DATA, but in application code.')
    print(f'  For all 130K entities, scale linearly (~{130000/(len(df)/elapsed):.0f}s).')
````

## Section 2: Scheduled Sync — Periodic Refresh

For features recomputed on a schedule (hourly, daily), run the backfill on a timer. Since Valkey has no native scheduled export, the pattern is **Cloud Scheduler → Cloud Run → pipelined HSET**.

### Freshness vs Complexity Tradeoff

| Approach | Freshness | Complexity | Cost |
|---|---|---|---|
| One-time backfill | Manual | Lowest | One-off compute |
| Scheduled sync | Minutes–hours | Low | Periodic compute |
| Continuous query (Section 3) | Seconds | Moderate | Always-on BQ slot + subscriber |
| Dual-write (Section 6) | Real-time | Higher | Application-managed |

### Cloud Run Job for Scheduled Sync

The scheduled sync job reads recomputed features from BigQuery and writes them to Valkey. Deploy as a Cloud Run job, trigger with Cloud Scheduler.

````python
scheduled_sync_job = f"""
# sync_job.py — deployed as a Cloud Run job, triggered by Cloud Scheduler

import redis, os
from google.cloud import bigquery

def run_sync():
    bq = bigquery.Client()
    client = redis.Redis(
        host=os.environ['REDIS_HOST'], port=int(os.environ['REDIS_PORT']),
        ssl=True, ssl_ca_certs=os.environ.get('REDIS_CERT'),
        decode_responses=True,
    )

    # Recompute features in BigQuery (e.g., rolling aggregates from events)
    query = f'''
        SELECT entity_group, entity_id,
               COUNT(*) AS event_count_24h,
               COUNTIF(event_type = 'purchase') AS purchases_24h,
               ROUND(AVG(metric_1), 4) AS avg_metric_1_24h
        FROM `{BQ_PROJECT}.{BQ_DATASET}.sparse_events`
        WHERE event_timestamp > TIMESTAMP_SUB(CURRENT_TIMESTAMP(), INTERVAL 24 HOUR)
        GROUP BY entity_group, entity_id
    '''
    df = bq.query(query).to_dataframe()

    pipe = client.pipeline(transaction=False)
    for i, (_, row) in enumerate(df.iterrows()):
        key = f"features:{{row['entity_group']}}:{{row['entity_id']}}"
        pipe.hset(key, mapping={{
            'event_count_24h': str(row['event_count_24h']),
            'purchases_24h': str(row['purchases_24h']),
            'avg_metric_1_24h': str(row['avg_metric_1_24h']),
        }})
        if (i + 1) % 500 == 0:
            pipe.execute(); pipe = client.pipeline(transaction=False)
    pipe.execute()
    print(f"Synced {{len(df)}} entities")

if __name__ == '__main__':
    run_sync()
"""
print('Cloud Run scheduled sync job:')
print(scheduled_sync_job)
````

## Section 3: Continuous Sync — BigQuery Continuous Query → Pub/Sub

For near-real-time sync, a BigQuery [continuous query](https://cloud.google.com/bigquery/docs/continuous-queries-introduction) captures new/changed rows via `APPENDS()` and publishes them to Pub/Sub. A subscriber pulls them and writes to Valkey.

```
BigQuery                    Pub/Sub                     Valkey
  dense_features  ----->  valkey-feature-sync  ----->  HSET features:*
  (source)               (topic)                       (destination)
  Continuous Query       Messages (JSON)               Python subscriber
  INSERT INTO topic()    one per row                   parses + writes
```

Unlike Bigtable (continuous query → `EXPORT DATA` to Bigtable) and Spanner (continuous query → `INSERT INTO` Spanner), Valkey isn't a valid continuous-query destination — so we route through **Pub/Sub** as the bridge.

````python
# Show the continuous query SQL
continuous_query_sql = f"""
-- BigQuery Continuous Query → Pub/Sub → Valkey
-- Run with: bq query --use_legacy_sql=false --continuous=true

INSERT INTO `{PROJECT_ID}.pubsub.topic({PUBSUB_TOPIC})`
SELECT TO_JSON_STRING(
    STRUCT(
        entity_group,
        entity_id,
        feature_int_1,
        feature_float_1,
        feature_float_2,
        feature_string_1,
        feature_bool_1,
        label,
        CURRENT_TIMESTAMP() AS sync_timestamp
    )
) AS data
FROM APPENDS(
    TABLE `{BQ_PROJECT}.{BQ_DATASET}.dense_features`,
    NULL  -- start from now
) AS t
"""

print('Continuous query SQL:')
print(continuous_query_sql)
````

### The Pub/Sub Subscriber — The Bridge

The subscriber pulls messages, parses JSON, and writes to Valkey with an idempotency check (skip stale messages by comparing `sync_timestamp`).

````python
def process_message(message_data, redis_client):
    """Process a single Pub/Sub message and write to Valkey."""
    data = json.loads(message_data)

    entity_group = data.pop('entity_group')
    entity_id = data.pop('entity_id')
    sync_timestamp = data.pop('sync_timestamp', None)

    key = f'features:{entity_group}:{entity_id}'

    # Idempotency check: only update if this message is newer
    existing_ts = redis_client.hget(key, '_sync_timestamp')
    if existing_ts and sync_timestamp and existing_ts >= sync_timestamp:
        return 'skipped'  # Stale message

    # Encode features
    mapping = {}
    for field, value in data.items():
        if value is not None:
            mapping[field] = str(value)
    if sync_timestamp:
        mapping['_sync_timestamp'] = sync_timestamp

    # Write to Valkey
    redis_client.hset(key, mapping=mapping)
    return 'written'

print('Subscriber function ready.')
````

### Simulate the Pipeline

Publish test messages, pull them, write to Valkey, and verify.

````python
# Publish test messages
# Use a SYNCTEST entity_group so we never overwrite real A-Z feature data.
test_entities = [
    {'entity_group': 'SYNCTEST', 'entity_id': '000001', 'feature_float_1': 99.99,
     'feature_string_1': 'synced_cat', 'sync_timestamp': time.strftime('%Y-%m-%dT%H:%M:%S')},
    {'entity_group': 'SYNCTEST', 'entity_id': '000002', 'feature_float_1': 88.88,
     'feature_string_1': 'synced_dog', 'sync_timestamp': time.strftime('%Y-%m-%dT%H:%M:%S')},
    {'entity_group': 'SYNCTEST', 'entity_id': '000003', 'feature_float_1': 77.77,
     'feature_string_1': 'synced_bird', 'sync_timestamp': time.strftime('%Y-%m-%dT%H:%M:%S')},
]

for entity in test_entities:
    msg = json.dumps(entity).encode('utf-8')
    future = publisher.publish(topic_path, msg)
    future.result()

print(f'Published {len(test_entities)} test messages to {PUBSUB_TOPIC}')
print(f'  (using entity_group=SYNCTEST so real feature data is untouched)')
time.sleep(2)  # Wait for delivery
````

````python
# Pull and process messages
# A single pull() may not return all just-published messages (Pub/Sub delivers
# asynchronously), so we pull in a short loop until we've drained the expected count.
if REDIS_AVAILABLE:
    results = {'written': 0, 'skipped': 0, 'error': 0}
    total_processed = 0
    expected = len(test_entities)
    deadline = time.time() + 15  # stop after 15s regardless

    while total_processed < expected and time.time() < deadline:
        response = subscriber.pull(subscription=subscription_path, max_messages=100)
        if not response.received_messages:
            time.sleep(1)
            continue

        ack_ids = []
        for msg in response.received_messages:
            try:
                status = process_message(msg.message.data.decode('utf-8'), decode_client)
                results[status] += 1
            except Exception as e:
                results['error'] += 1
                print(f'  Error: {e}')
            ack_ids.append(msg.ack_id)
            total_processed += 1

        if ack_ids:
            subscriber.acknowledge(subscription=subscription_path, ack_ids=ack_ids)

    print(f'Processed {total_processed} messages:')
    print(f'  Written: {results["written"]}')
    print(f'  Skipped: {results["skipped"]}')
    print(f'  Errors:  {results["error"]}')
````

````python
# Verify synced data in Valkey
if REDIS_AVAILABLE:
    for entity in test_entities:
        key = f'features:{entity["entity_group"]}:{entity["entity_id"]}'
        result = decode_client.hgetall(key)
        synced_float = result.get('feature_float_1', 'missing')
        synced_string = result.get('feature_string_1', 'missing')
        print(f'  {key}: float_1={synced_float}, string_1={synced_string}')

    # Cleanup: remove the SYNCTEST keys so they don't linger in the dataset
    for entity in test_entities:
        decode_client.delete(f'features:{entity["entity_group"]}:{entity["entity_id"]}')
    print(f'\nCleaned up {len(test_entities)} SYNCTEST keys.')
````

### Batch Subscriber with Pipelining

For production throughput, batch multiple HSET commands into a Redis pipeline before acknowledging.

````python
def batch_subscriber(subscription_path, redis_client, max_messages=100, pipeline_size=50):
    """Pull messages in batch, write to Redis with pipelining."""
    response = subscriber.pull(
        subscription=subscription_path,
        max_messages=max_messages,
    )

    if not response.received_messages:
        return 0

    pipe = redis_client.pipeline(transaction=False)
    ack_ids = []
    count = 0

    for msg in response.received_messages:
        try:
            data = json.loads(msg.message.data.decode('utf-8'))
            entity_group = data.pop('entity_group')
            entity_id = data.pop('entity_id')
            data.pop('sync_timestamp', None)

            key = f'features:{entity_group}:{entity_id}'
            mapping = {k: str(v) for k, v in data.items() if v is not None}
            pipe.hset(key, mapping=mapping)
            count += 1

            if count % pipeline_size == 0:
                pipe.execute()
                pipe = redis_client.pipeline(transaction=False)

        except Exception as e:
            print(f'  Error processing message: {e}')

        ack_ids.append(msg.ack_id)

    pipe.execute()  # Flush remaining

    if ack_ids:
        subscriber.acknowledge(subscription=subscription_path, ack_ids=ack_ids)

    return count

print('Batch subscriber function ready.')
print('  Uses Redis pipelining to batch multiple HSET commands.')
print('  Acknowledges all messages after successful write.')
````

## Section 4: Subscriber Deployment *(Valkey-specific)*

Bigtable uses managed Dataflow for streaming sync. Valkey's custom subscriber has to **run somewhere**. Two models:

| Model | Mechanism | Best for |
|---|---|---|
| **Pull subscription** | Subscriber polls `subscriber.pull()` in a loop | Cloud Run service / GKE, full control over batching |
| **Push subscription** | Pub/Sub POSTs to an HTTP endpoint | Cloud Run service that scales to zero, simpler ops |

### Pull Subscriber on Cloud Run

A long-running Cloud Run service (min-instances=1) that continuously pulls and writes to Valkey with pipelining. This gives you control over batch size and pipeline flushing.

````python
pull_subscriber_service = '''
# subscriber_service.py — Cloud Run service, min-instances=1

import redis, os, json
from google.cloud import pubsub_v1
from concurrent.futures import TimeoutError

subscriber = pubsub_v1.SubscriberClient()
sub_path = subscriber.subscription_path(os.environ['PROJECT'], os.environ['SUBSCRIPTION'])
client = redis.Redis(host=os.environ['REDIS_HOST'], port=int(os.environ['REDIS_PORT']),
                     ssl=True, ssl_ca_certs=os.environ.get('REDIS_CERT'), decode_responses=True)

def callback(message):
    try:
        data = json.loads(message.data.decode('utf-8'))
        eg, eid = data.pop('entity_group'), data.pop('entity_id')
        data.pop('sync_timestamp', None)
        key = f'features:{eg}:{eid}'
        client.hset(key, mapping={k: str(v) for k, v in data.items() if v is not None})
        message.ack()
    except Exception:
        message.nack()  # redelivery, eventually dead-letter

# Streaming pull with flow control (backpressure)
flow = pubsub_v1.types.FlowControl(max_messages=500)
future = subscriber.subscribe(sub_path, callback=callback, flow_control=flow)
try:
    future.result()
except (KeyboardInterrupt, TimeoutError):
    future.cancel()
'''
print('Pull subscriber (Cloud Run service):')
print(pull_subscriber_service)
````

````python
push_subscription_setup = f"""
# Push model: Pub/Sub POSTs messages to a Cloud Run HTTP endpoint.
# The service scales to zero when idle — no always-on subscriber.

gcloud pubsub subscriptions create {PUBSUB_SUBSCRIPTION}-push \\
    --topic {PUBSUB_TOPIC} \\
    --push-endpoint "https://valkey-sync-xxxx.run.app/sync" \\
    --push-auth-service-account sync-sa@{PROJECT_ID}.iam.gserviceaccount.com \\
    --ack-deadline 60

# The Cloud Run endpoint decodes the message, HSETs to Valkey, returns 200 to ack.
# 4xx/5xx responses trigger redelivery (and eventually dead-letter).
"""
print('Push subscription setup:')
print(push_subscription_setup)
print('Trade-off: push scales to zero (cheaper when idle) but gives less control over batching.')
print('Pull keeps a warm connection pool to Valkey and batches with pipelines (higher throughput).')
````

## Section 5: TTL-on-Sync — Automatic Staleness Bounds *(Valkey-specific)*

This is a sync pattern **unique to Valkey**. When the subscriber writes a feature, it also sets an `EXPIRE`. If the sync pipeline stalls (continuous query stops, subscriber crashes), stale features **automatically disappear** instead of being served indefinitely.

This gives you an automatic staleness guarantee that neither Bigtable nor Spanner has — there, stale data sits until overwritten. With TTL-on-sync, a feature that isn't refreshed within its TTL window is evicted, and your serving layer can treat a missing key as "stale, fall back to BigQuery."

**Pattern:** set TTL = (sync interval × safety factor). If you sync every 60s, set TTL to 180s — a missed sync round still leaves the feature live, but a sustained outage expires it.

````python
if REDIS_AVAILABLE:
    SYNC_INTERVAL = 60        # seconds between sync rounds
    TTL_SAFETY_FACTOR = 3     # tolerate 3 missed rounds before expiring
    SYNC_TTL = SYNC_INTERVAL * TTL_SAFETY_FACTOR

    def sync_with_ttl(redis_client, entity_group, entity_id, features, ttl=SYNC_TTL):
        """Write features with an expiry so stale data self-evicts."""
        key = f'features:{entity_group}:{entity_id}'
        pipe = redis_client.pipeline(transaction=True)
        pipe.hset(key, mapping={k: str(v) for k, v in features.items() if v is not None})
        pipe.expire(key, ttl)
        pipe.execute()

    # Demo: sync a feature with TTL
    sync_with_ttl(decode_client, 'TEST', 'TTL001',
                  {'feature_float_1': 42.0, 'feature_string_1': 'fresh'})

    key = 'features:TEST:TTL001'
    print(f'TTL-on-sync demo:')
    print(f'  Sync interval:  {SYNC_INTERVAL}s')
    print(f'  TTL:            {SYNC_TTL}s (tolerates {TTL_SAFETY_FACTOR} missed rounds)')
    print(f'  Key {key} TTL:  {decode_client.ttl(key)}s remaining')
    print(f'\nIf the sync pipeline stalls, this key expires after {SYNC_TTL}s.')
    print(f'  Serving layer treats missing key as "stale" -> fall back to BigQuery.')

    decode_client.delete(key)
````

### Serving-Side Staleness Handling

With TTL-on-sync, the serving layer gets staleness detection for free: a missing key means the feature is stale (or never synced). The fallback reads from BigQuery and optionally re-populates Valkey.

````python
if REDIS_AVAILABLE:
    def read_feature_with_fallback(redis_client, entity_group, entity_id):
        """Read from Valkey; if expired/missing, fall back to BigQuery."""
        key = f'features:{entity_group}:{entity_id}'
        data = redis_client.hgetall(key)
        if data:
            return data, 'valkey'  # fresh

        # Stale or missing -> fall back to BigQuery
        df = bq.query(f'''
            SELECT feature_float_1, feature_string_1, label
            FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
            WHERE entity_group = '{entity_group}' AND entity_id = '{entity_id}'
        ''').to_dataframe()
        if len(df):
            row = df.iloc[0].to_dict()
            # Re-populate Valkey with a fresh TTL (serving subset)
            redis_client.hset(key, mapping={k: str(v) for k, v in row.items() if pd.notna(v)})
            redis_client.expire(key, SYNC_TTL)
            return row, 'bigquery_fallback'
        return None, 'not_found'

    # Case 1: fresh key already in Valkey (from NB1 load)
    data, source = read_feature_with_fallback(decode_client, 'A', '000001')
    print(f'A/000001 (in Valkey):     served from {source}')

    # Case 2: simulate an expired key to trigger the fallback.
    # Save the original hash first so we can restore it with no side effects.
    demo_key = 'features:C:000050'
    saved = decode_client.hgetall(demo_key)
    decode_client.delete(demo_key)

    data, source = read_feature_with_fallback(decode_client, 'C', '000050')
    print(f'C/000050 (was expired):   served from {source}')
    print(f'  -> fallback read it from BigQuery and re-populated Valkey with a fresh TTL')

    data, source = read_feature_with_fallback(decode_client, 'C', '000050')
    print(f'C/000050 (re-populated):  served from {source}')

    # Restore the original full hash (no lasting side effects, no stray TTL)
    decode_client.delete(demo_key)
    if saved:
        decode_client.hset(demo_key, mapping=saved)
        decode_client.persist(demo_key)
    print(f'\nRestored {demo_key} to its original {len(saved)} fields (no TTL).')
````

## Section 6: Dual-Write — Real-Time Path

For features that change in real time (not just batch-recomputed), the application writes directly to Valkey **and** to BigQuery. Valkey gets the update instantly (sub-ms); BigQuery gets it for the offline store and training.

This mirrors Bigtable's dual-write pattern. Two variants:

| Variant | Path | Consistency |
|---|---|---|
| **App → Pub/Sub → both** | One publish fans out to Valkey subscriber + BigQuery subscriber | Eventual, both targets symmetric |
| **App → Valkey + BigQuery** | App writes both directly | App owns consistency; Valkey first for low latency |

For real-time serving features (counters, session state), write Valkey first (the serving path), then BigQuery asynchronously (the durable record).

````python
if REDIS_AVAILABLE:
    def dual_write(redis_client, bq_client, entity_group, entity_id, features):
        """Write to Valkey (serving) first, then BigQuery (durable) async."""
        # 1. Valkey - instant, the serving path
        key = f'features:{entity_group}:{entity_id}'
        redis_client.hset(key, mapping={k: str(v) for k, v in features.items()})

        # 2. BigQuery - durable record (streaming insert)
        # In production this is batched/async; shown inline for clarity
        table = f'{BQ_PROJECT}.{BQ_DATASET}.realtime_writes'
        rows = [{'entity_group': entity_group, 'entity_id': entity_id, **features}]
        # bq_client.insert_rows_json(table, rows)  # uncomment with a real target table
        return key

    # Demo: dual-write a real-time feature
    start = time.perf_counter()
    key = dual_write(decode_client, bq, 'TEST', 'DUAL001',
                     {'live_score': 0.95, 'updated': 'now'})
    valkey_ms = (time.perf_counter() - start) * 1000

    print(f'Dual-write pattern:')
    print(f'  Valkey write (serving path): {valkey_ms:.2f}ms')
    print(f'  BigQuery write: batched async (not blocking serving)')
    print(f'  Verified in Valkey: {decode_client.hgetall(key)}')
    print(f'\nValkey-first ordering keeps the serving path sub-ms;')
    print(f'  BigQuery gets the durable copy for training without blocking reads.')

    decode_client.delete(key)
````

## Section 7: The Propagation Race

Compare end-to-end latency across the sync patterns. How long from "data changes in BigQuery" to "feature readable in Valkey"?

````python
if REDIS_AVAILABLE:
    # Measure end-to-end sync latency
    sync_key = 'features:TEST:SYNC001'
    decode_client.delete(sync_key)  # Clean start

    # Publish
    t_publish = time.perf_counter()
    msg = json.dumps({
        'entity_group': 'TEST', 'entity_id': 'SYNC001',
        'feature_float_1': 12345.6789,
        'latency_test': 'true',
    }).encode('utf-8')
    publisher.publish(topic_path, msg).result()
    publish_ms = (time.perf_counter() - t_publish) * 1000

    # Pull until our message arrives (Pub/Sub delivery is asynchronous, so a single
    # pull may miss it). We measure from publish to first successful read.
    t_pull_start = time.perf_counter()
    write_ms = None
    deadline = time.time() + 15
    while time.time() < deadline:
        response = subscriber.pull(subscription=subscription_path, max_messages=10)
        if not response.received_messages:
            continue
        ack_ids = [m.ack_id for m in response.received_messages]
        got_it = False
        for m in response.received_messages:
            data = json.loads(m.message.data.decode('utf-8'))
            if data.get('entity_group') == 'TEST' and data.get('entity_id') == 'SYNC001':
                t_write = time.perf_counter()
                key = f'features:{data["entity_group"]}:{data["entity_id"]}'
                mapping = {k: str(v) for k, v in data.items()
                           if k not in ('entity_group', 'entity_id') and v is not None}
                decode_client.hset(key, mapping=mapping)
                write_ms = (time.perf_counter() - t_write) * 1000
                got_it = True
        subscriber.acknowledge(subscription=subscription_path, ack_ids=ack_ids)
        if got_it:
            break
    delivery_ms = (time.perf_counter() - t_pull_start) * 1000

    # Verify
    result = decode_client.hget(sync_key, 'feature_float_1')

    print(f'Sync propagation latency (publish -> readable in Valkey):')
    print(f'  Publish to Pub/Sub:    {publish_ms:>8.0f} ms')
    print(f'  Delivery + pull:       {delivery_ms:>8.0f} ms  (Pub/Sub async delivery)')
    print(f'  Write to Valkey:       {write_ms:>8.1f} ms' if write_ms is not None else '  Write to Valkey:        (message not received)')
    print(f'  Verified in Valkey:    feature_float_1 = {result}')
    print(f'\nWith a real continuous query, end-to-end delivery is typically a few seconds.')

    # Cleanup
    decode_client.delete(sync_key)
````

````python
if REDIS_AVAILABLE:
    # Compare the propagation characteristics of each pattern
    print('Sync pattern propagation comparison:\n')
    print(f'{"Pattern":<28s} {"Freshness":<18s} {"Path"}')
    rows = [
        ('One-time backfill', 'Manual', 'BQ read -> pipeline HSET'),
        ('Scheduled sync', 'Minutes-hours', 'Scheduler -> Cloud Run -> HSET'),
        ('Continuous query', 'Seconds', 'CQ -> Pub/Sub -> subscriber -> HSET'),
        ('Dual-write', 'Real-time (<1ms)', 'App -> Valkey directly'),
    ]
    for name, fresh, path in rows:
        print(f'{name:<28s} {fresh:<18s} {path}')

    print(f'\nThe right pattern depends on how fresh the feature needs to be:')
    print(f'  - Static/slow features (demographics)  -> scheduled sync')
    print(f'  - Recomputed features (rolling aggs)   -> continuous query')
    print(f'  - Real-time signals (counters, state)  -> dual-write')
````

## Section 8: Memory Pressure During Bulk Sync *(Valkey-specific)*

Bulk sync loads data into RAM — and Valkey has a hard memory ceiling. A full backfill of a large dataset can trigger eviction if it exceeds `maxmemory`, especially when combined with existing data.

Things Bigtable/Spanner never have to worry about (they scale disk automatically), but matter here:
- **Watch `used_memory` during sync** — don't let a backfill OOM the instance
- **Eviction interplay** — with `volatile-lru`, a backfill of TTL keys can evict older TTL keys mid-sync
- **Sizing** — ensure (existing data + backfill) fits with headroom

````python
if REDIS_AVAILABLE:
    mem = decode_client.info('memory')
    stats = decode_client.info('stats')
    maxmem = int(mem.get('maxmemory', 0))
    used = mem['used_memory']

    print(f'Memory state (monitor this during bulk sync):')
    print(f'  used_memory:        {mem["used_memory_human"]}')
    print(f'  maxmemory:          {mem.get("maxmemory_human", "(instance limit)")}')
    if maxmem > 0:
        print(f'  headroom:           {(maxmem-used)/1024/1024:.0f} MB ({100*used/maxmem:.1f}% used)')
    print(f'  evicted_keys:       {stats.get("evicted_keys", 0):,}')

    # NOTE: Managed Memorystore disables the CONFIG command. The maxmemory-policy
    # is set at the instance level (Terraform redis_configs / gcloud), not via
    # CONFIG SET. INFO still works, so monitor memory through INFO, not CONFIG.
    try:
        policy = decode_client.config_get("maxmemory-policy")["maxmemory-policy"]
        print(f'  maxmemory_policy:   {policy}')
    except redis.ResponseError:
        # Try reading it from INFO memory (some versions expose it there)
        policy = mem.get('maxmemory_policy', '(set at instance level)')
        print(f'  maxmemory_policy:   {policy}  [CONFIG disabled on managed Memorystore]')

    # Estimate backfill cost before running it
    sample_key = decode_client.randomkey()
    if sample_key:
        per_key = decode_client.memory_usage(sample_key) or 2048
        est_full = per_key * 130000 / 1024 / 1024
        print(f'\nEst. bytes/entity:  {per_key:,}')
        print(f'  Est. 130K backfill: {est_full:.0f} MB')
        print(f'\nBefore a full backfill, confirm headroom > estimated size.')
        print(f'  If using TTL keys + volatile-lru, a backfill can evict older TTL keys.')
````

## Section 9: BigQuery Reservations & Batch-Read Analogs

### Continuous Query Reservations

Even though Valkey isn't a BigQuery destination, the **continuous query that feeds Pub/Sub runs in BigQuery** and needs a CONTINUOUS-assignment reservation (BigQuery Enterprise edition). This is the same requirement as Bigtable/Spanner continuous-query sync.

```
CREATE RESERVATION -> assign with job_type = CONTINUOUS -> run the continuous query
```

### Bigtable Sync Features Without Valkey Analogs

Two Bigtable sync capabilities don't map directly to Valkey:

| Bigtable feature | Purpose | Valkey equivalent |
|---|---|---|
| **Data Boost** | Serverless batch reads for training without impacting serving | Read training data **straight from BigQuery** (the offline store) — Valkey is serving-only, so there's no read-isolation problem to solve |
| **Continuous materialized views** | Pre-computed aggregations inside Bigtable | Compute aggregates in a **BigQuery materialized view**, then sync the result through the Pub/Sub bridge |

Valkey is simpler here precisely because it's an in-memory **serving** layer — batch/training reads never touch it; they go to BigQuery directly.

## Section 10: Error Handling, Idempotency & Dead-Letter Topics

Production sync needs:
- **Retry**: transient errors (Redis timeout) → `nack()` for redelivery with backoff
- **Idempotency**: duplicate messages → timestamp check before overwriting (shown in Section 3's `process_message`)
- **Dead-letter**: messages that fail repeatedly → move to a dead-letter topic for inspection

````python
dead_letter_config = f"""
# Create a dead-letter topic for messages that fail processing

gcloud pubsub topics create {PUBSUB_TOPIC}-dead-letter \\
    --project={PROJECT_ID}

# Update subscription with dead-letter policy
gcloud pubsub subscriptions update {PUBSUB_SUBSCRIPTION} \\
    --dead-letter-topic={PUBSUB_TOPIC}-dead-letter \\
    --max-delivery-attempts=5 \\
    --project={PROJECT_ID}

# Messages that fail 5 times are moved to the dead-letter topic
# for manual inspection and reprocessing.
"""

print('Dead-letter topic configuration:')
print(dead_letter_config)
````
