# Valkey Feature Store — Vector Search with HNSW

Valkey's killer feature for ML serving: native HNSW approximate nearest neighbor search via `FT.CREATE` and `FT.SEARCH`. Sub-5ms KNN on 130K+ vectors — the only production-grade ANN in this feature store series.

**What you'll learn:**
- Store embeddings as packed float32 bytes in Redis hashes
- Create FLAT index (brute-force) and HNSW index (approximate nearest neighbor)
- KNN query with `FT.SEARCH` and `DIALECT 2`
- Distance metrics: COSINE, L2, IP (inner product)
- HNSW parameters: M, EF_CONSTRUCTION, EF_RUNTIME — tuning accuracy vs latency
- Benchmark: FLAT vs HNSW at scale
- Comparison across stores: all now have native vector search; differ on index type, pre-filtering, and latency

**Why this matters:** Recommendation engines need to retrieve top-K from large candidate sets in real time. Valkey HNSW does this in <5ms at 130K vectors. Bigtable's brute-force takes 100ms+. Spanner and BigQuery have native vector search too, but with higher latency — Valkey is the pick when sub-ms real-time retrieval is the requirement.

**Prerequisites:** Run [Environment](./Valkey%20Feature%20Store%20-%20Environment.ipynb) and [Fundamentals](./Valkey%20Feature%20Store%20-%20Fundamentals.ipynb) first.

> **🔄 Redis vs Valkey — connecting:** This series runs on **Memorystore for Valkey** (Cluster Mode Disabled), so we read the instance endpoint via `gcloud memorystore instances describe` and connect with the standard `redis.Redis` client. On **Memorystore for Redis** you'd instead use the `google-cloud-redis` admin client (`redis_v1.CloudRedisClient`) to get `host`/`port`. The client code and all commands below are identical either way — Valkey is Redis-protocol compatible. (Cluster Mode Enabled would use `redis.cluster.RedisCluster` against a discovery endpoint.)

```python
import redis
import numpy as np
import struct
import time
import os
import matplotlib.pyplot as plt

def wait_for_index(client, index_name, expected_docs, timeout=120):
    """Poll FT.INFO until indexing finishes (or num_docs reaches expected)."""
    deadline = time.time() + timeout
    while time.time() < deadline:
        info_list = client.execute_command('FT.INFO', index_name)
        info = dict(zip(info_list[::2], info_list[1::2]))
        num_docs = int(info.get('num_docs', 0))
        indexing = str(info.get('indexing', '0'))
        if num_docs >= expected_docs and indexing in ('0', 'False', 'false'):
            return num_docs
        time.sleep(1)
    return num_docs

# --- Connect to Memorystore for Valkey (Cluster Mode Disabled -> redis.Redis) ---
import subprocess as _sp, json as _json
def _valkey_endpoint_and_cert():
    r = _sp.run(['gcloud','memorystore','instances','describe', VALKEY_INSTANCE,
                 '--location', REGION, '--project', PROJECT_ID, '--format','json'],
                capture_output=True, text=True)
    if r.returncode != 0:
        return None, None, None
    inst = _json.loads(r.stdout)
    host = port = None
    for ep in inst.get('endpoints', []):
        for conn in ep.get('connections', []):
            psc = conn.get('pscAutoConnection', {})
            if psc.get('ipAddress'):
                host, port = psc['ipAddress'], int(psc.get('port', 6379)); break
        if host: break
    if not host:
        for de in inst.get('discoveryEndpoints', []):
            if de.get('address'):
                host, port = de['address'], int(de.get('port', 6379)); break
    # CA cert for TLS
    cert_path = os.path.join(os.path.expanduser('~'), '.valkey-certs', f'{VALKEY_INSTANCE}-ca.pem')
    if not os.path.exists(cert_path):
        cr = _sp.run(['gcloud','memorystore','instances','get-certificate-authority', VALKEY_INSTANCE,
                      '--location', REGION, '--project', PROJECT_ID, '--format','json'],
                     capture_output=True, text=True)
        if cr.returncode == 0:
            ca = _json.loads(cr.stdout); pem = ''
            for c1 in ca.get('managedServerCa', {}).get('caCerts', []):
                for c2 in c1.get('certificates', []):
                    pem += c2 + '\n'
            if pem:
                os.makedirs(os.path.dirname(cert_path), exist_ok=True)
                open(cert_path, 'w').write(pem)
            else:
                cert_path = None
        else:
            cert_path = None
    return host, port, cert_path

try:
    REDIS_HOST, REDIS_PORT, CERT_PATH = _valkey_endpoint_and_cert()
    if not REDIS_HOST:
        raise RuntimeError('Valkey instance not found — run the Environment notebook first.')
    connect_kwargs = dict(host=REDIS_HOST, port=REDIS_PORT)
    if CERT_PATH:
        connect_kwargs['ssl'] = True
        connect_kwargs['ssl_ca_certs'] = CERT_PATH
        connect_kwargs['ssl_check_hostname'] = False  # cert is for a DNS name, not the PSC IP
    decode_client = redis.Redis(decode_responses=True, **connect_kwargs)
    bytes_client = redis.Redis(decode_responses=False, **connect_kwargs)
    info = decode_client.info('server')
    print(f'Valkey: {REDIS_HOST}:{REDIS_PORT}')
    print(f'Engine: {info.get("server_name", "valkey")} {info.get("valkey_version", info["redis_version"])} (redis-compat {info["redis_version"]})')
    REDIS_AVAILABLE = True
except Exception as e:
    print(f'Valkey not available: {e}')
    print('For local dev: docker run -p 6379:6379 valkey/valkey:8')
    REDIS_AVAILABLE = False
    print(f'Redis not available: {e}')
```

## 1. Store Embeddings as Packed Bytes

FT.SEARCH requires embeddings stored as raw float32 bytes (not JSON strings). Each 64-dim embedding is 256 bytes.

```python
if REDIS_AVAILABLE:
    # We tag the embedding load with a marker so we can detect (and replace) the
    # earlier non-clustered embeddings if they're present.
    marker = bytes_client.get(b'emb:__clustered__')
    if marker == b'1':
        print('Clustered embeddings already loaded.')
        SKIP_EMB_LOAD = True
    else:
        # Drop any existing emb:* keys (old random embeddings) before reloading.
        existing = list(bytes_client.scan_iter(match=b'emb:*', count=1000))
        if existing:
            print(f'Removing {len(existing):,} existing embeddings (non-clustered) before reload...')
            for j in range(0, len(existing), 1000):
                bytes_client.delete(*existing[j:j+1000])
        SKIP_EMB_LOAD = False
        print('Loading clustered embeddings...')
```

```python
if REDIS_AVAILABLE and not SKIP_EMB_LOAD:
    np.random.seed(42)
    categories = ['electronics', 'clothing', 'food', 'sports', 'books']

    # Give the embeddings real CLUSTER STRUCTURE so nearest-neighbor is meaningful.
    # Pure random vectors are near-equidistant in high dimensions (curse of
    # dimensionality), which makes "top-K nearest" almost arbitrary and recall
    # uninformative. Real embeddings cluster — so we model that: each vector is a
    # cluster centroid + small noise. This is what makes ANN recall measurable.
    N_CLUSTERS = 50
    centroids = np.random.randn(N_CLUSTERS, EMBEDDING_DIM).astype(np.float32)
    centroids /= np.linalg.norm(centroids, axis=1, keepdims=True)
    NOISE = 0.15  # small relative to centroid separation -> tight, separable clusters

    n_loaded = 0
    overall_start = time.perf_counter()

    for group_code in range(65, 91):  # A-Z
        group = chr(group_code)
        pipe = bytes_client.pipeline(transaction=False)

        for i in range(1, 5001):
            eid = f'{i:06d}'
            key = f'emb:{group}:{eid}'

            # Assign this entity to a cluster, then perturb the centroid slightly
            cluster = (group_code * 5000 + i) % N_CLUSTERS
            emb = centroids[cluster] + NOISE * np.random.randn(EMBEDDING_DIM).astype(np.float32)
            emb = emb / np.linalg.norm(emb)
            emb_bytes = emb.tobytes()

            category = categories[(group_code * 1000 + i) % len(categories)]
            score = str(round(np.random.uniform(0, 100), 2))

            pipe.hset(key, mapping={
                b'embedding': emb_bytes,
                b'entity_group': group.encode(),
                b'entity_id': eid.encode(),
                b'category': category.encode(),
                b'score': score.encode(),
            })

            if i % 500 == 0:
                pipe.execute()
                pipe = bytes_client.pipeline(transaction=False)

        pipe.execute()
        n_loaded += 5000
        if group_code % 5 == 0:
            elapsed = time.perf_counter() - overall_start
            print(f'  Loaded through group {group}: {n_loaded:,} embeddings ({elapsed:.0f}s)')

    elapsed = time.perf_counter() - overall_start
    print(f'\nLoad complete: {n_loaded:,} embeddings ({EMBEDDING_DIM}-dim, {N_CLUSTERS} clusters) in {elapsed:.0f}s')
    print(f'  Per embedding: {EMBEDDING_DIM * 4} bytes (float32)')
    print(f'  Cluster structure makes nearest-neighbor meaningful -> measurable recall.')
    bytes_client.set(b'emb:__clustered__', b'1')
```

---
## 2. Create FLAT Index (Brute-Force)

FLAT indexes scan every vector — exact results but O(n) latency.

```python
if REDIS_AVAILABLE:
    # Drop existing index if present
    try:
        decode_client.execute_command('FT.DROPINDEX', 'idx:flat')
    except Exception:
        pass

    # Create FLAT (brute-force) index
    start = time.perf_counter()
    decode_client.execute_command(
        'FT.CREATE', 'idx:flat',
        'ON', 'HASH',
        'PREFIX', '1', 'emb:',
        'SCHEMA',
        'embedding', 'VECTOR', 'FLAT', '6',
        'TYPE', 'FLOAT32',
        'DIM', str(EMBEDDING_DIM),
        'DISTANCE_METRIC', 'COSINE',
        'entity_group', 'TAG',
        'category', 'TAG',
        'score', 'NUMERIC',
    )
    flat_create_ms = (time.perf_counter() - start) * 1000

    print(f'FLAT index created: {flat_create_ms:.0f}ms')
    print(f'  Scans all vectors for exact results')

    # Wait until indexing actually completes (not a fixed sleep) — important after
    # a fresh 130K-vector reload, so downstream recall/queries see a full index.
    n = wait_for_index(decode_client, 'idx:flat', expected_docs=130000)
    print(f'  Indexed: {n:,} docs')
```

---
## 3. Create HNSW Index (Approximate Nearest Neighbor)

HNSW (Hierarchical Navigable Small World) provides sub-linear search time with tunable accuracy.

```python
if REDIS_AVAILABLE:
    try:
        decode_client.execute_command('FT.DROPINDEX', 'idx:hnsw')
    except Exception:
        pass

    start = time.perf_counter()
    decode_client.execute_command(
        'FT.CREATE', 'idx:hnsw',
        'ON', 'HASH',
        'PREFIX', '1', 'emb:',
        'SCHEMA',
        'embedding', 'VECTOR', 'HNSW', '10',
        'TYPE', 'FLOAT32',
        'DIM', str(EMBEDDING_DIM),
        'DISTANCE_METRIC', 'COSINE',
        'M', '16',
        'EF_CONSTRUCTION', '200',
        'entity_group', 'TAG',
        'category', 'TAG',
        'score', 'NUMERIC',
    )
    hnsw_create_ms = (time.perf_counter() - start) * 1000

    print(f'HNSW index created: {hnsw_create_ms:.0f}ms')
    print(f'  M=16, EF_CONSTRUCTION=200')
    print(f'  Sub-linear search with tunable accuracy')

    # Wait until indexing actually completes (HNSW build over 130K vectors takes
    # longer than a fixed sleep — poll FT.INFO so recall is measured on a full index).
    n = wait_for_index(decode_client, 'idx:hnsw', expected_docs=130000)
    print(f'  Indexed: {n:,} docs')
```

---
## 4. KNN Query with FT.SEARCH

Search for the K nearest neighbors to a query vector.

```python
if REDIS_AVAILABLE:
    # Read a query embedding
    query_bytes = bytes_client.hget(b'emb:A:000001', b'embedding')
    query_vec = np.frombuffer(query_bytes, dtype=np.float32)
    print(f'Query vector: {query_vec.shape} (entity A/000001)')

    K = 10

    # FLAT search (exact)
    start = time.perf_counter()
    flat_results = bytes_client.execute_command(
        'FT.SEARCH', b'idx:flat',
        b'*=>[KNN ' + str(K).encode() + b' @embedding $query_vec AS distance]',
        b'PARAMS', b'2', b'query_vec', query_bytes,
        b'RETURN', b'3', b'entity_group', b'entity_id', b'distance',
        b'DIALECT', b'2',
    )
    flat_ms = (time.perf_counter() - start) * 1000

    # HNSW search (approximate)
    start = time.perf_counter()
    hnsw_results = bytes_client.execute_command(
        'FT.SEARCH', b'idx:hnsw',
        b'*=>[KNN ' + str(K).encode() + b' @embedding $query_vec AS distance]',
        b'PARAMS', b'2', b'query_vec', query_bytes,
        b'RETURN', b'3', b'entity_group', b'entity_id', b'distance',
        b'DIALECT', b'2',
    )
    hnsw_ms = (time.perf_counter() - start) * 1000

    print(f'\nKNN search ({K} neighbors):')
    print(f'  FLAT (exact):  {flat_ms:.1f}ms')
    print(f'  HNSW (approx): {hnsw_ms:.1f}ms')
    print(f'  Speedup:       {flat_ms / hnsw_ms:.1f}x')

    # Parse HNSW results, sorting by distance client-side (KNN result order is not
    # guaranteed without SORTBY, which this RediSearch version rejects).
    n_results = hnsw_results[0]
    parsed = []
    for i in range(1, min(n_results * 2 + 1, len(hnsw_results)), 2):
        fields = hnsw_results[i + 1]
        fd = dict(zip(fields[::2], fields[1::2]))
        parsed.append((
            fd.get(b'entity_group', b'?').decode(),
            fd.get(b'entity_id', b'?').decode(),
            float(fd.get(b'distance', b'inf')),
        ))
    parsed.sort(key=lambda x: x[2])
    print(f'\nTop {K} nearest neighbors (HNSW):')
    for group, eid, dist in parsed:
        print(f'  {group}/{eid}: distance = {dist:.4f}')
```

## 5. Latency Benchmark — FLAT vs HNSW

```python
if REDIS_AVAILABLE:
    n_queries = 50
    flat_latencies = []
    hnsw_latencies = []

    for i in range(n_queries):
        # Random query vector
        q = np.random.randn(EMBEDDING_DIM).astype(np.float32)
        q = q / np.linalg.norm(q)
        q_bytes = q.tobytes()

        # FLAT
        start = time.perf_counter()
        bytes_client.execute_command(
            'FT.SEARCH', b'idx:flat',
            b'*=>[KNN 10 @embedding $q AS d]',
            b'PARAMS', b'2', b'q', q_bytes,
            b'RETURN', b'1', b'd', b'DIALECT', b'2',
        )
        flat_latencies.append((time.perf_counter() - start) * 1000)

        # HNSW
        start = time.perf_counter()
        bytes_client.execute_command(
            'FT.SEARCH', b'idx:hnsw',
            b'*=>[KNN 10 @embedding $q AS d]',
            b'PARAMS', b'2', b'q', q_bytes,
            b'RETURN', b'1', b'd', b'DIALECT', b'2',
        )
        hnsw_latencies.append((time.perf_counter() - start) * 1000)

    print(f'Vector search latency ({n_queries} queries, 130K vectors, {EMBEDDING_DIM}-dim):\n')
    print(f'{"Index":.<15s} {"Median":>10s} {"p95":>10s} {"Mean":>10s}')
    for name, lats in [('FLAT', flat_latencies), ('HNSW', hnsw_latencies)]:
        s = sorted(lats)
        print(f'{name:.<15s} {s[n_queries//2]:>8.1f}ms {s[int(n_queries*0.95)]:>8.1f}ms {sum(lats)/len(lats):>8.1f}ms')
```

```python
if REDIS_AVAILABLE:
    fig, ax = plt.subplots(figsize=(10, 5))

    bp = ax.boxplot([flat_latencies, hnsw_latencies], labels=['FLAT (brute-force)', 'HNSW (ANN)'],
                    patch_artist=True)
    bp['boxes'][0].set_facecolor('#EA4335')
    bp['boxes'][0].set_alpha(0.7)
    bp['boxes'][1].set_facecolor('#34A853')
    bp['boxes'][1].set_alpha(0.7)

    ax.set_ylabel('Latency (ms)', fontweight='bold')
    ax.set_title('Vector Search: FLAT vs HNSW (130K vectors)', fontsize=13, fontweight='bold')
    ax.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.show()
```

## 6. HNSW Parameters — Tuning Accuracy vs Latency

HNSW has three tuning knobs. On **managed Memorystore**, all three are set at **index creation** (`FT.CREATE`) — the managed search does not accept a per-query `EF_RUNTIME` override (supported `FT.SEARCH` params are `NOCONTENT`, `TIMEOUT`, `PARAMS`, `RETURN`, `LIMIT`, `DIALECT`).

| Parameter | Set at | Effect |
|---|---|---|
| `M` | FT.CREATE | Graph connectivity per layer — higher = better recall, more memory |
| `EF_CONSTRUCTION` | FT.CREATE | Build-time quality — higher = better recall, slower build |
| `EF_RUNTIME` | FT.CREATE (default search breadth) | Higher = better recall, slower search |

To change `EF_RUNTIME` on Memorystore you rebuild the index with a new default. On self-managed Redis/Valkey you can additionally override it per query.

```python
if REDIS_AVAILABLE:
    # NOTE: Memorystore's managed search does NOT accept a per-query EF_RUNTIME
    # override (the supported FT.SEARCH params are NOCONTENT, TIMEOUT, PARAMS,
    # RETURN, LIMIT, DIALECT). EF_RUNTIME and M/EF_CONSTRUCTION are set at INDEX
    # CREATION time (FT.CREATE) instead. So here we measure HNSW vs FLAT latency
    # at the index's configured settings, rather than sweeping EF at query time.
    q = np.random.randn(EMBEDDING_DIM).astype(np.float32)
    q = q / np.linalg.norm(q)
    q_bytes = q.tobytes()

    def time_search(index, n=30):
        lats = []
        for _ in range(n):
            start = time.perf_counter()
            bytes_client.execute_command(
                'FT.SEARCH', index,
                b'*=>[KNN 10 @embedding $q AS d]',
                b'PARAMS', b'2', b'q', q_bytes,
                b'RETURN', b'1', b'd', b'DIALECT', b'2',
            )
            lats.append((time.perf_counter() - start) * 1000)
        return sorted(lats)[len(lats)//2]

    hnsw_med = time_search(b'idx:hnsw')
    flat_med = time_search(b'idx:flat')

    print('HNSW vs FLAT at configured index settings (median of 30 queries):\n')
    print(f'  HNSW (M=16, EF_CONSTRUCTION=200):  {hnsw_med:.2f} ms')
    print(f'  FLAT (exact brute-force):          {flat_med:.2f} ms')
    print(f'  Speedup: {flat_med / hnsw_med:.1f}x')

    print('\nWhere the HNSW tuning knobs live:')
    print('  M               (FT.CREATE) graph connectivity — higher = better recall, more memory')
    print('  EF_CONSTRUCTION (FT.CREATE) build quality — higher = better recall, slower build')
    print('  EF_RUNTIME      (FT.CREATE) default search breadth — higher = better recall, slower search')
    print('\nOn self-managed Redis/Valkey you can also pass EF_RUNTIME per query;')
    print('on managed Memorystore it is fixed at index creation, so you tune it by')
    print('recreating the index with a different EF_RUNTIME default.')
```

## 7. Hybrid Filtered Search — Metadata Filter + KNN

The defining capability of `FT.SEARCH`: combine a **metadata filter** (TAG/NUMERIC fields) with **vector KNN** in a single query. The index already declares `category TAG` and `score NUMERIC`, so we can restrict candidates to (say) electronics with score > 50 and rank *those* by vector distance — all server-side, one round-trip.

Bigtable's KNN can't pre-filter (row key only). Spanner *can* combine ANN with SQL WHERE — its strength — but at higher latency; Valkey does filter + KNN in one sub-5ms call.

```python
if REDIS_AVAILABLE:
    # Hybrid search: combine a metadata filter (TAG/NUMERIC) with KNN in one query.
    # The index already declares `category TAG` and `score NUMERIC`, so we can
    # pre-filter candidates and rank by vector distance in a single FT.SEARCH.
    query_bytes = bytes_client.hget(b'emb:A:000001', b'embedding')
    K = 10

    # Pure KNN (no filter) — searches all 130K vectors
    pure = bytes_client.execute_command(
        'FT.SEARCH', b'idx:hnsw',
        b'*=>[KNN ' + str(K).encode() + b' @embedding $q AS distance]',
        b'PARAMS', b'2', b'q', query_bytes,
        b'RETURN', b'2', b'category', b'distance',
        b'DIALECT', b'2',
    )

    # Hybrid: only electronics with score > 50, ranked by vector distance
    hybrid = bytes_client.execute_command(
        'FT.SEARCH', b'idx:hnsw',
        b'(@category:{electronics} @score:[50 +inf])=>[KNN ' + str(K).encode() + b' @embedding $q AS distance]',
        b'PARAMS', b'2', b'q', query_bytes,
        b'RETURN', b'3', b'category', b'score', b'distance',
        b'DIALECT', b'2',
    )

    print(f'Pure KNN (no filter): {pure[0]} candidates considered (all vectors)')
    print(f'Hybrid (category=electronics AND score>50):\n')
    n = hybrid[0]
    print(f'  Matches: {n}')
    rows = []
    for i in range(1, min(n*2+1, len(hybrid)), 2):
        f = dict(zip(hybrid[i+1][::2], hybrid[i+1][1::2]))
        rows.append((
            f.get(b'category', b'?').decode(),
            f.get(b'score', b'?').decode(),
            float(f.get(b'distance', b'inf')),
        ))
    rows.sort(key=lambda x: x[2])  # nearest first (client-side sort)
    for cat, sc, d in rows:
        print(f'  {cat:<12s} score={sc:<7s} distance={d:.4f}')

    print('\nFilter syntax (DIALECT 2):')
    print('  @tag_field:{value}      TAG equality')
    print('  @num_field:[min max]    NUMERIC range ([50 +inf] = >= 50)')
    print('  (filter)=>[KNN k @vec $q]   filter FIRST, then KNN on the matches')
    print('\nThis is the FT.SEARCH defining capability: metadata filter + vector rank')
    print('in ONE sub-5ms call. Bigtable KNN cannot pre-filter; Spanner combines ANN+SQL but at higher latency.')
```

## 8. Recall@K — Measuring HNSW Accuracy

HNSW is **approximate** — it trades exactness for speed and may miss some true neighbors. Section 6 covered the latency side; here we measure the **accuracy** side. Using FLAT (exact brute-force) as ground truth, we compute recall@K for the HNSW index as configured — the fraction of the true top-K that HNSW also returns.

```python
if REDIS_AVAILABLE:
    # Recall@K: HNSW is APPROXIMATE and may miss some true neighbors. We measure
    # recall by treating FLAT (exact brute-force) as ground truth and checking how
    # many of HNSW's top-K it recovers. We query with REAL entity embeddings (not
    # random vectors) so the queries land near actual data — the realistic case.
    import numpy as np
    K = 10
    N_QUERIES = 30

    def knn_ids(index, qbytes, k):
        res = bytes_client.execute_command(
            'FT.SEARCH', index,
            b'*=>[KNN ' + str(k).encode() + b' @embedding $q AS d]',
            b'PARAMS', b'2', b'q', qbytes,
            b'RETURN', b'0', b'DIALECT', b'2',
        )
        # With RETURN 0, FT.SEARCH returns [count, key1, key2, ...]
        return set(res[1:res[0] + 1]) if res and res[0] else set()

    # Use real stored embeddings as queries (sample N_QUERIES entities)
    rng = np.random.default_rng(0)
    sample_ids = [(chr(65 + int(rng.integers(0, 26))), f'{int(rng.integers(1, 5001)):06d}')
                  for _ in range(N_QUERIES)]
    queries = []
    for g, e in sample_ids:
        qb = bytes_client.hget(f'emb:{g}:{e}'.encode(), b'embedding')
        if qb:
            queries.append(qb)

    total_recall = 0.0
    for q in queries:
        truth = knn_ids(b'idx:flat', q, K)       # exact neighbors
        approx = knn_ids(b'idx:hnsw', q, K)      # approximate neighbors
        if truth:
            total_recall += len(truth & approx) / len(truth)
    recall = total_recall / len(queries)

    print(f'Recall@{K} of HNSW vs FLAT ground truth ({len(queries)} real-entity queries):\n')
    print(f'  Recall@{K}: {recall:.1%}')
    print(f'  (fraction of the exact top-{K} that HNSW also returned)')

    print('\nInterpreting recall:')
    print('  - Recall near 100% means HNSW is finding almost all true neighbors.')
    print('  - Lower recall trades accuracy for speed; raise it by rebuilding the')
    print('    index with a higher EF_RUNTIME (or EF_CONSTRUCTION / M).')
    print('  - For most feature-store retrieval, 95%+ recall at a few ms is the goal.')
    print('\nNote: recall is only meaningful when embeddings have real structure.')
    print('Random vectors are near-equidistant in high dimensions (curse of')
    print('dimensionality), so nearest-neighbor is ill-defined — the clustered')
    print('embeddings loaded in this notebook make this measurement valid.')
```

## 9. Combining Vectors with Scalar Features

In a real feature store the embedding lives in the **same entity hash** as the scalar features. Retrieval is two-stage: HNSW finds candidate keys by vector similarity, then a pipelined read pulls the full feature set (vector + scalars) for ranking. This is the exact pattern the trading capstone (NB8) uses.

```python
if REDIS_AVAILABLE:
    # Vectors + scalar features in one entity: the embedding lives in the SAME hash
    # as the scalar features. A two-stage retrieval reads both in the serving path:
    #   Stage 1: HNSW finds candidate entity keys by vector similarity
    #   Stage 2: pipeline-read the full feature hash (vector + scalars) for each
    query_bytes = bytes_client.hget(b'emb:A:000001', b'embedding')
    K = 5

    # Stage 1: retrieve candidate keys via HNSW
    res = bytes_client.execute_command(
        'FT.SEARCH', b'idx:hnsw',
        b'*=>[KNN ' + str(K).encode() + b' @embedding $q AS distance]',
        b'PARAMS', b'2', b'q', query_bytes,
        b'RETURN', b'1', b'distance', b'DIALECT', b'2',
    )
    n = res[0]
    # Pair each key with its distance and sort client-side (order not guaranteed
    # without SORTBY). res = [count, key1, [fields1], key2, [fields2], ...]
    cand = []
    for i in range(1, min(n*2+1, len(res)), 2):
        fd = dict(zip(res[i+1][::2], res[i+1][1::2]))
        cand.append((res[i], float(fd.get(b'distance', b'inf'))))
    cand.sort(key=lambda x: x[1])
    candidate_keys = [k for k, _ in cand]

    # Stage 2: pipeline-read scalar features for the candidates (from the emb: hash)
    pipe = bytes_client.pipeline(transaction=False)
    for k in candidate_keys:
        pipe.hmget(k, b'category', b'score', b'entity_group', b'entity_id')
    rows = pipe.execute()

    print(f'Two-stage retrieval — vector candidates + scalar features:\n')
    print(f'{"entity":<14s} {"category":<12s} {"score":>8s}')
    for k, vals in zip(candidate_keys, rows):
        cat = (vals[0] or b'?').decode()
        sc = (vals[1] or b'?').decode()
        eg = (vals[2] or b'?').decode()
        eid = (vals[3] or b'?').decode()
        print(f'{eg+"/"+eid:<14s} {cat:<12s} {sc:>8s}')

    print('\nThe pattern: HNSW returns candidate KEYS, then a pipelined HMGET pulls')
    print('the scalar features for ranking/serving. Vector + scalars co-located in')
    print('one hash means one round-trip per candidate (batched via pipeline).')
    print('This is exactly the two-stage retrieval the trading capstone (NB8) uses.')
```

## 10. Comparison Across Feature Store Approaches

All four serving stores now offer **native** vector search — the differences are in index type, latency, and how you pre-filter.

| Approach | Method | Index | Pre-Filtering | Latency (130K) |
|---|---|---|---|---|
| **Valkey** | `FT.SEARCH` KNN | HNSW (ANN) or FLAT (exact) | TAG / NUMERIC filter in the same query | ~2–5ms (HNSW) |
| **Bigtable** | GoogleSQL `COSINE_DISTANCE()` | Brute-force | Row key prefix only | ~100–300ms |
| **Spanner** | `VECTOR_SEARCH` / distance functions | Vector index (ScaNN, ANN) or exact KNN | **Full SQL WHERE** (any indexed column) | ~10–50ms |
| **BigQuery** | `VECTOR_SEARCH` | Vector index (IVF / TreeAH) | SQL predicate pushdown | ~100ms–seconds |

**How to choose:**
- **Valkey** — lowest latency for real-time retrieval; HNSW gives sub-5ms ANN at scale. Pre-filtering limited to indexed TAG/NUMERIC fields.
- **Spanner** — native ANN **plus** arbitrary SQL pre-filtering and JOINs in the same query; the strongest "filter then search" story when candidates are narrowed by rich predicates.
- **BigQuery** — best for batch/offline vector work and analytics-scale corpora; native index, but query-engine latency.
- **Bigtable** — brute-force only (no ANN index); fine for small candidate sets reached by row key.

**For sub-millisecond real-time serving, Valkey HNSW is the pick.** When vector search is one step inside a richer SQL query (filter by many columns, JOIN other features), Spanner's native ANN + SQL is the more flexible choice.
