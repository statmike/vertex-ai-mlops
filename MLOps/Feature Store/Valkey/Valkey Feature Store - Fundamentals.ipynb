# Valkey Feature Store — Fundamentals

First sync from BigQuery via Pub/Sub subscriber. Store features as Redis hashes. Measure sub-millisecond read latency.

**What you'll learn:**
- Build a sync pipeline: BigQuery → Pub/Sub → Python subscriber → HSET to Valkey
- Hash data structure: `HSET features:A:000001 feature_float_1 42.5 ...`
- Read patterns: `HGET` (single field), `HMGET` (multiple fields), `HGETALL` (all fields)
- Key naming convention: `features:{entity_group}:{entity_id}`
- Point-lookup latency: <1ms
- Pipelining for bulk initial load
- Two-client pattern: string client + bytes client
- Comparison to Bigtable and Spanner read latency

**Key difference from Bigtable:** Bigtable uses `EXPORT DATA` (one SQL command) for bulk sync. Valkey has no native BigQuery export — you build the pipeline yourself. The trade-off: more code, but sub-millisecond reads.

**Prerequisites:** Run [Environment](./Valkey%20Feature%20Store%20-%20Environment.ipynb) first.

> **🔄 Redis vs Valkey — connecting:** This series runs on **Memorystore for Valkey** (Cluster Mode Disabled), so we read the instance endpoint via `gcloud memorystore instances describe` and connect with the standard `redis.Redis` client. On **Memorystore for Redis** you'd instead use the `google-cloud-redis` admin client (`redis_v1.CloudRedisClient`) to get `host`/`port`. The client code and all commands below are identical either way — Valkey is Redis-protocol compatible. (Cluster Mode Enabled would use `redis.cluster.RedisCluster` against a discovery endpoint.)

```python
from google.cloud import bigquery
import redis
import pandas as pd
import numpy as np
import time
import json
import os
import matplotlib.pyplot as plt

bq = bigquery.Client(project=PROJECT_ID)

# --- Connect to Memorystore for Valkey (Cluster Mode Disabled -> redis.Redis) ---
import subprocess as _sp, json as _json
def _valkey_endpoint_and_cert():
    r = _sp.run(['gcloud','memorystore','instances','describe', VALKEY_INSTANCE,
                 '--location', REGION, '--project', PROJECT_ID, '--format','json'],
                capture_output=True, text=True)
    if r.returncode != 0:
        return None, None, None
    inst = _json.loads(r.stdout)
    host = port = None
    for ep in inst.get('endpoints', []):
        for conn in ep.get('connections', []):
            psc = conn.get('pscAutoConnection', {})
            if psc.get('ipAddress'):
                host, port = psc['ipAddress'], int(psc.get('port', 6379)); break
        if host: break
    if not host:
        for de in inst.get('discoveryEndpoints', []):
            if de.get('address'):
                host, port = de['address'], int(de.get('port', 6379)); break
    # CA cert for TLS
    cert_path = os.path.join(os.path.expanduser('~'), '.valkey-certs', f'{VALKEY_INSTANCE}-ca.pem')
    if not os.path.exists(cert_path):
        cr = _sp.run(['gcloud','memorystore','instances','get-certificate-authority', VALKEY_INSTANCE,
                      '--location', REGION, '--project', PROJECT_ID, '--format','json'],
                     capture_output=True, text=True)
        if cr.returncode == 0:
            ca = _json.loads(cr.stdout); pem = ''
            for c1 in ca.get('managedServerCa', {}).get('caCerts', []):
                for c2 in c1.get('certificates', []):
                    pem += c2 + '\n'
            if pem:
                os.makedirs(os.path.dirname(cert_path), exist_ok=True)
                open(cert_path, 'w').write(pem)
            else:
                cert_path = None
        else:
            cert_path = None
    return host, port, cert_path

try:
    REDIS_HOST, REDIS_PORT, CERT_PATH = _valkey_endpoint_and_cert()
    if not REDIS_HOST:
        raise RuntimeError('Valkey instance not found — run the Environment notebook first.')
    connect_kwargs = dict(host=REDIS_HOST, port=REDIS_PORT)
    if CERT_PATH:
        connect_kwargs['ssl'] = True
        connect_kwargs['ssl_ca_certs'] = CERT_PATH
        connect_kwargs['ssl_check_hostname'] = False  # cert is for a DNS name, not the PSC IP
    decode_client = redis.Redis(decode_responses=True, **connect_kwargs)
    bytes_client = redis.Redis(decode_responses=False, **connect_kwargs)
    info = decode_client.info('server')
    print(f'Valkey: {REDIS_HOST}:{REDIS_PORT}')
    print(f'Engine: {info.get("server_name", "valkey")} {info.get("valkey_version", info["redis_version"])} (redis-compat {info["redis_version"]})')
    REDIS_AVAILABLE = True
except Exception as e:
    print(f'Valkey not available: {e}')
    print('For local dev: docker run -p 6379:6379 valkey/valkey:8')
    REDIS_AVAILABLE = False
```

## 1. Bulk Load — BigQuery to Valkey

Load all 130K entities from BigQuery into Valkey using pipelined HSET commands. Pipelining batches multiple commands into a single network roundtrip.

```python
# Feature columns to load
FEATURE_COLUMNS = [
    'entity_group', 'entity_id',
    'feature_int_1', 'feature_int_2', 'feature_int_3', 'feature_int_4', 'feature_int_5',
    'feature_float_1', 'feature_float_2', 'feature_float_3', 'feature_float_4', 'feature_float_5',
    'feature_string_1', 'feature_string_2', 'feature_string_3',
    'feature_bool_1', 'feature_bool_2', 'feature_bool_3',
    'label',
]

# Add bulk float columns
FEATURE_COLUMNS += [f'feature_float_{i}' for i in range(6, 31)]  # First 25 bulk floats

print(f'Loading {len(FEATURE_COLUMNS)} columns per entity')
```

```python
if REDIS_AVAILABLE:
    # Check if data already loaded
    existing_keys = decode_client.dbsize()
    if existing_keys >= 130000:
        print(f'Data already loaded: {existing_keys:,} keys in Redis.')
        SKIP_LOAD = True
    else:
        print(f'Current keys: {existing_keys:,}. Loading...')
        SKIP_LOAD = False
else:
    SKIP_LOAD = True
    print('Redis not available — skipping load.')
```

```python
if not SKIP_LOAD:
    cols_sql = ', '.join(FEATURE_COLUMNS)
    groups = [chr(c) for c in range(65, 91)]  # A-Z
    pipeline_batch_size = 500
    total_loaded = 0
    overall_start = time.perf_counter()

    for group in groups:
        group_start = time.perf_counter()

        df = bq.query(f'''
            SELECT {cols_sql}
            FROM `{BQ_PROJECT}.{BQ_DATASET}.dense_features`
            WHERE entity_group = '{group}'
            ORDER BY entity_id
        ''').to_dataframe()

        # Pipeline HSET commands
        pipe = decode_client.pipeline(transaction=False)
        for idx, (_, row) in enumerate(df.iterrows()):
            key = f'features:{row["entity_group"]}:{row["entity_id"]}'
            mapping = {}
            for col in FEATURE_COLUMNS[2:]:  # Skip entity_group, entity_id
                val = row[col]
                if pd.isna(val):
                    continue
                mapping[col] = str(val)
            pipe.hset(key, mapping=mapping)

            if (idx + 1) % pipeline_batch_size == 0:
                pipe.execute()
                pipe = decode_client.pipeline(transaction=False)

        pipe.execute()  # Flush remaining

        elapsed = time.perf_counter() - group_start
        total_loaded += len(df)
        print(f'  Group {group}: {len(df):,} entities in {elapsed:.1f}s ({total_loaded:,} total)')

    overall_elapsed = time.perf_counter() - overall_start
    print(f'\nLoad complete: {total_loaded:,} entities in {overall_elapsed:.0f}s')
    print(f'  Pipeline batch size: {pipeline_batch_size}')
    print(f'  Throughput: {total_loaded / overall_elapsed:,.0f} entities/s')
```

```python
# Verify load
if REDIS_AVAILABLE:
    key_count = decode_client.dbsize()
    print(f'Total keys in Redis: {key_count:,}')

    # Check a sample entity
    sample = decode_client.hgetall('features:A:000001')
    print(f'\nSample entity (features:A:000001):')
    print(f'  Fields: {len(sample)}')
    for field in ['feature_int_1', 'feature_float_1', 'feature_string_1', 'feature_bool_1', 'label']:
        print(f'  {field}: {sample.get(field, "(missing)")}')

    mem = decode_client.info('memory')
    print(f'\nMemory usage:')
    print(f'  Used: {mem["used_memory_human"]}')
    print(f'  Peak: {mem["used_memory_peak_human"]}')
```

## 2. Read Patterns — HGET, HMGET, HGETALL

Redis hashes support three read patterns with different tradeoffs.

```python
if REDIS_AVAILABLE:
    key = 'features:A:000001'

    # HGET — single field
    start = time.perf_counter()
    val = decode_client.hget(key, 'feature_float_1')
    hget_ms = (time.perf_counter() - start) * 1000
    print(f'HGET (1 field):    {hget_ms:.3f}ms  →  feature_float_1 = {val}')

    # HMGET — multiple fields
    fields = ['feature_float_1', 'feature_int_1', 'feature_string_1', 'label']
    start = time.perf_counter()
    vals = decode_client.hmget(key, fields)
    hmget_ms = (time.perf_counter() - start) * 1000
    print(f'HMGET (4 fields):  {hmget_ms:.3f}ms  →  {dict(zip(fields, vals))}')

    # HGETALL — all fields
    start = time.perf_counter()
    all_vals = decode_client.hgetall(key)
    hgetall_ms = (time.perf_counter() - start) * 1000
    print(f'HGETALL ({len(all_vals)} fields): {hgetall_ms:.3f}ms')
```

## 3. Point Read Latency Benchmark

Measure sub-millisecond read latency across many entities.

```python
if REDIS_AVAILABLE:
    n_reads = 1000
    latencies = []

    for i in range(n_reads):
        group = chr(65 + (i % 26))
        eid = f'{(i * 37 % 5000) + 1:06d}'
        key = f'features:{group}:{eid}'

        start = time.perf_counter()
        decode_client.hgetall(key)
        latencies.append((time.perf_counter() - start) * 1000)

    print(f'Point read latency (HGETALL, {n_reads} reads):\n')
    print(f'  Min:    {min(latencies):.3f} ms')
    print(f'  Median: {sorted(latencies)[n_reads//2]:.3f} ms')
    print(f'  p95:    {sorted(latencies)[int(n_reads*0.95)]:.3f} ms')
    print(f'  p99:    {sorted(latencies)[int(n_reads*0.99)]:.3f} ms')
    print(f'  Max:    {max(latencies):.3f} ms')
    print(f'  Mean:   {sum(latencies)/len(latencies):.3f} ms')
```

```python
if REDIS_AVAILABLE:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Histogram
    axes[0].hist(latencies, bins=50, color='#EA4335', alpha=0.8, edgecolor='white')
    axes[0].axvline(sorted(latencies)[n_reads//2], color='black', linestyle='--',
                    label=f'Median: {sorted(latencies)[n_reads//2]:.3f}ms')
    axes[0].set_xlabel('Latency (ms)', fontweight='bold')
    axes[0].set_ylabel('Count', fontweight='bold')
    axes[0].set_title('Point Read Latency Distribution', fontsize=13, fontweight='bold')
    axes[0].legend()

    # Time series
    axes[1].plot(latencies[:200], alpha=0.7, linewidth=0.5, color='#4285F4')
    axes[1].set_xlabel('Read #', fontweight='bold')
    axes[1].set_ylabel('Latency (ms)', fontweight='bold')
    axes[1].set_title('Latency Over Time (first 200)', fontsize=13, fontweight='bold')
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

## 4. Pipelined Batch Reads

Read multiple entities in a single network roundtrip using Redis pipelining.

```python
if REDIS_AVAILABLE:
    batch_sizes = [1, 5, 10, 50, 100, 500]

    print(f'Pipelined batch read latency:\n')
    print(f'{"Entities":>10s} {"Total":>10s} {"Per-entity":>12s}')

    batch_results = []
    for n in batch_sizes:
        keys = [f'features:{chr(65 + i % 26)}:{(i * 37 % 5000) + 1:06d}' for i in range(n)]

        latencies = []
        for _ in range(5):
            pipe = decode_client.pipeline(transaction=False)
            for key in keys:
                pipe.hgetall(key)

            start = time.perf_counter()
            results = pipe.execute()
            latencies.append((time.perf_counter() - start) * 1000)

        median = sorted(latencies)[2]
        per_entity = median / n
        batch_results.append((n, median, per_entity))
        print(f'{n:>10,} {median:>8.2f}ms {per_entity:>10.3f}ms')

    print(f'\nPipelining amortizes network roundtrip across all entities.')
```

```python
if REDIS_AVAILABLE:
    fig, ax = plt.subplots(figsize=(10, 5))

    sizes = [r[0] for r in batch_results]
    totals = [r[1] for r in batch_results]

    ax.plot(sizes, totals, 'o-', color='#4285F4', linewidth=2, markersize=8)
    for s, t in zip(sizes, totals):
        ax.annotate(f'{t:.1f}ms', (s, t), textcoords='offset points', xytext=(0, 12),
                    ha='center', fontsize=9)

    ax.set_xlabel('Batch Size (entities)', fontweight='bold')
    ax.set_ylabel('Total Latency (ms)', fontweight='bold')
    ax.set_title('Pipelined Batch Read Latency', fontsize=13, fontweight='bold')
    ax.set_xscale('log')
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

## 5. Comparison to Bigtable and Spanner

| Operation | Valkey | Bigtable | Spanner |
|---|---|---|---|
| Single entity (HGETALL) | <1ms | ~3–5ms | ~5–10ms |
| 10 entities (pipelined) | ~1–2ms | ~10–20ms | ~10–15ms |
| 100 entities (pipelined) | ~3–5ms | ~50–100ms | ~15–30ms |
| Sync from BigQuery | Custom (Pub/Sub bridge) | Native (EXPORT DATA) | Native (continuous query) |
| Storage | In-memory (RAM) | Disk (SSD) | Disk (SSD) |
| Data types | Strings/bytes | Bytes | Native typed |
| Max capacity | 300GB RAM | Petabytes | Petabytes |

**Valkey is 5–10× faster for reads** but requires all data to fit in memory and has no native BigQuery sync path.
