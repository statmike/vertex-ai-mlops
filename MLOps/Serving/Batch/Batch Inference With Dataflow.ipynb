# Batch Inference With Dataflow

Run batch inference using [Apache Beam](https://beam.apache.org/) on [Dataflow](https://cloud.google.com/dataflow). Models load directly into Dataflow workers — no container build, no endpoint. Pre/post processing are native Beam pipeline stages that scale independently.

This notebook demonstrates:

- Custom `ModelHandler` for HuggingFace models stored in GCS
- Pre/post processing as Beam `DoFn` stages
- Dynamic batching with `BatchElements`
- Local testing with `DirectRunner` before deploying to Dataflow
- Multi-model inference with separate pipeline branches
- Orchestrating Dataflow jobs with Vertex AI Pipelines (KFP)

**How it works:** Apache Beam's `RunInference` transform manages model loading, batching, and inference. Each Dataflow worker downloads the model from GCS once at startup, then processes elements in batches. Dataflow autoscales workers based on data volume.

**Story arc:**
1. Build a custom `ModelHandler` that loads HuggingFace models from GCS
2. Test the pipeline locally with `DirectRunner`
3. Run on Dataflow with `DataflowRunner`
4. Compare two models (DistilBERT vs BERT) in a multi-model pipeline
5. Orchestrate the workflow as a KFP pipeline

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params) — returns `POSITIVE`/`NEGATIVE`
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params) — returns `LABEL_0`/`LABEL_1`

Download two sentiment analysis models from HuggingFace and upload them to GCS. Dataflow workers will download these model files at pipeline startup — no container build or Model Registry upload needed.

- **v1 (DistilBERT):** Smaller, faster — primary model
- **v2 (BERT):** Larger — second model for multi-model comparison

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

## Custom ModelHandler

Apache Beam's `RunInference` transform uses a `ModelHandler` to manage model loading and inference. The handler is instantiated once per worker — the model downloads from GCS at worker startup, then serves all elements routed to that worker.

Our handler wraps HuggingFace's `pipeline()` API, which handles tokenization and inference internally. Input elements are dicts with at least a `text` field; the handler extracts text, runs batch inference, and returns `PredictionResult` objects that carry both the original element and the model output.

````python
class SentimentModelHandler(ModelHandler):
    """ModelHandler that loads a HuggingFace sentiment model from GCS."""

    def __init__(self, model_gcs_uri):
        self._model_gcs_uri = model_gcs_uri
        self._model_name = model_gcs_uri.split('/')[-1]  # 'v1' or 'v2'

    def load_model(self):
        """Download model from GCS and load with HuggingFace pipeline.
        Called once per worker at startup."""
        from transformers import pipeline as hf_pipeline
        from google.cloud import storage as gcs_lib

        client = gcs_lib.Client()
        parts = self._model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]

        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)

        return hf_pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)

    def run_inference(self, batch, model, inference_args=None):
        """Run sentiment analysis on a batch of elements.

        Args:
            batch: List of dicts, each with at least a 'text' field.
            model: The HuggingFace pipeline loaded by load_model().

        Returns:
            List of PredictionResult with the original element and prediction dict.
        """
        texts = [element['text'] for element in batch]
        results = model(texts)
        return [
            PredictionResult(example=element, inference=result)
            for element, result in zip(batch, results)
        ]

    def batch_elements_kwargs(self):
        """Configure dynamic batching bounds for RunInference."""
        return {'min_batch_size': 1, 'max_batch_size': 32}

print('SentimentModelHandler defined')
````

### Pre/Post Processing DoFns

Pre- and post-processing run as separate Beam stages that scale independently from inference:

- **`PreprocessFn`** — Clean text, validate, pass metadata through
- **`PostprocessFn`** — Extract prediction fields, format for BigQuery output

````python
class PreprocessFn(beam.DoFn):
    """Clean and validate text before inference."""
    def process(self, row):
        text = row.get('text', '')
        if isinstance(text, str):
            text = text.strip()
        if text:
            yield {
                'id': row.get('id', 0),
                'text': text,
                'category': row.get('category', ''),
            }


class PostprocessFn(beam.DoFn):
    """Format PredictionResult for BigQuery output."""
    def __init__(self, model_name=''):
        self._model_name = model_name

    def process(self, prediction_result):
        element = prediction_result.example
        result = prediction_result.inference
        yield {
            'id': element['id'],
            'text': element['text'],
            'label': result['label'],
            'score': round(result['score'], 6),
            'model': self._model_name,
        }

print('DoFns defined: PreprocessFn, PostprocessFn')
````

## Test with DirectRunner

Run the pipeline locally to verify the model handler and DoFns work before deploying to Dataflow. `DirectRunner` executes everything in-process — no cloud resources needed.

````python
# Build model handler for DistilBERT (v1)
v1_handler = SentimentModelHandler(MODELS['v1']['artifact_uri'])

# Run locally with DirectRunner — write results to a temp file
DIRECT_OUTPUT = '/tmp/beam_direct_results'

with beam.Pipeline(options=PipelineOptions()) as p:
    (
        p
        | 'CreateInput' >> beam.Create([
            {'id': i + 1, 'text': s, 'category': c}
            for i, (s, c) in enumerate(zip(SENTENCES, categories))
        ])
        | 'Preprocess' >> beam.ParDo(PreprocessFn())
        | 'Infer' >> RunInference(model_handler=v1_handler)
        | 'Postprocess' >> beam.ParDo(PostprocessFn('distilbert'))
        | 'ToJson' >> beam.Map(json.dumps)
        | 'Write' >> beam.io.WriteToText(
            DIRECT_OUTPUT, file_name_suffix='.jsonl', shard_name_template='',
        )
    )

# Read results back
direct_results = [json.loads(line) for line in open(f'{DIRECT_OUTPUT}.jsonl') if line.strip()]
print(f'DirectRunner results ({len(direct_results)} sentences):')
````

## Pre/Post Processing Patterns

The pipeline above uses separate `DoFn` stages for pre- and post-processing. Apache Beam's `RunInference` also supports **inline** pre/post processing via `with_preprocess_fn()` and `with_postprocess_fn()`. Here's how the two approaches compare:

### Inline Functions (Alternative)

```python
def preprocess(element):
    return {**element, 'text': element['text'].strip()}

def postprocess(result):
    return {
        'id': result.example['id'],
        'label': result.inference['label'],
        'score': result.inference['score'],
    }

# Attach directly to RunInference
(
    p
    | ReadFromBigQuery(...)
    | RunInference(handler)
        .with_preprocess_fn(preprocess)
        .with_postprocess_fn(postprocess)
    | WriteToBigQuery(...)
)
```

### When to Use Each

| Approach | Scaling | Complexity | Best For |
|----------|---------|------------|----------|
| **Separate DoFns** | Each stage scales independently | More pipeline stages | Complex transforms, filtering, fan-out |
| **Inline functions** | Co-located with inference | Fewer stages, simpler | Lightweight transforms (field extraction, renaming) |

For production pipelines with non-trivial pre/post processing (SQL joins, external lookups, validation with dead-letter output), separate DoFns are the standard pattern. Inline functions work well for simple field mapping.

### Dynamic Batching

`RunInference` automatically batches elements before calling `run_inference()`. The `batch_elements_kwargs()` method on the handler controls batching bounds:

```python
def batch_elements_kwargs(self):
    return {'min_batch_size': 1, 'max_batch_size': 32}
```

Beam's `BatchElements` transform measures per-element processing time and adjusts batch size within these bounds to maximize throughput. Unlike Vertex AI Batch Prediction's fixed `batchSize` parameter, this adapts at runtime.

| Setting | Effect |
|---------|--------|
| `min_batch_size=1` | Don't wait for a full batch — process immediately if available |
| `max_batch_size=32` | Cap batch size to avoid OOM in the model |
| `max_batch_duration_secs` | Optional timeout — flush partial batches after N seconds |

**Rule of thumb:** Set `max_batch_size` based on your model's memory footprint. For transformer models, 16-32 is typical. Beam handles the rest.

## Multi-Model Inference

Run both DistilBERT and BERT on the same data in a single pipeline. Each model gets its own `RunInference` stage with a separate handler. Results are flattened into a single output with a `model` column identifying which model produced each prediction.

This pattern enables model comparison, A/B analysis, and ensemble scoring — all in one pipeline execution.

````python
# Multi-model pipeline with DirectRunner
v1_handler = SentimentModelHandler(MODELS['v1']['artifact_uri'])
v2_handler = SentimentModelHandler(MODELS['v2']['artifact_uri'])

MULTI_OUTPUT = '/tmp/beam_multi_results'

input_data = [
    {'id': i + 1, 'text': s, 'category': c}
    for i, (s, c) in enumerate(zip(SENTENCES, categories))
]

with beam.Pipeline(options=PipelineOptions()) as p:
    preprocessed = (
        p
        | 'Create' >> beam.Create(input_data)
        | 'Preprocess' >> beam.ParDo(PreprocessFn())
    )

    # Branch 1: DistilBERT
    distilbert = (
        preprocessed
        | 'InferV1' >> RunInference(model_handler=v1_handler)
        | 'PostV1' >> beam.ParDo(PostprocessFn('distilbert'))
    )

    # Branch 2: BERT
    bert = (
        preprocessed
        | 'InferV2' >> RunInference(model_handler=v2_handler)
        | 'PostV2' >> beam.ParDo(PostprocessFn('bert'))
    )

    # Merge and write results
    merged = (distilbert, bert) | 'Flatten' >> beam.Flatten()
    (
        merged
        | 'ToJson' >> beam.Map(json.dumps)
        | 'Write' >> beam.io.WriteToText(
            MULTI_OUTPUT, file_name_suffix='.jsonl', shard_name_template='',
        )
    )

# Read results back
multi_results = [json.loads(line) for line in open(f'{MULTI_OUTPUT}.jsonl') if line.strip()]
print(f'Multi-model results: {len(multi_results)} predictions ({len(multi_results)//2} per model)')
````

````python
# Display side-by-side comparison
print('Model Comparison — Same Sentences, Different Models')
print('=' * 70)
print()

# Group by sentence ID
by_id = {}
for r in multi_results:
    by_id.setdefault(r['id'], {})[r['model']] = r

for sid in sorted(by_id.keys()):
    preds = by_id[sid]
    text = list(preds.values())[0]['text']
    print(f'  "{text[:65]}..."')
    if 'distilbert' in preds:
        d = preds['distilbert']
        print(f'    DistilBERT: {d["label"]:>10} (score: {d["score"]:.6f})')
    if 'bert' in preds:
        b = preds['bert']
        print(f'    BERT:       {b["label"]:>10} (score: {b["score"]:.6f})')
    print()
````

### Multi-Model Patterns

| Pattern | How It Works | Best For |
|---------|-------------|----------|
| **Separate branches** (shown above) | Fan out from shared input, each branch has its own handler | Different model architectures, clear separation |
| **KeyedModelHandler** | Route elements to handlers by key in a single `RunInference` step | Same architecture, different weights (A/B testing) |
| **Sequential pipelines** | Run the full pipeline twice with different handlers | Simplest, but reads input data twice |

The **separate branches** pattern is most flexible — each branch can have its own pre/post processing, and the pipeline graph is easy to reason about. `KeyedModelHandler` (available in Beam 2.50+) is more efficient when you have many models of the same type:

```python
from apache_beam.ml.inference.base import KeyedModelHandler, KeyModelMapping

keyed_handler = KeyedModelHandler([
    KeyModelMapping(keys=['distilbert'], handler=v1_handler),
    KeyModelMapping(keys=['bert'], handler=v2_handler),
])
```

## Orchestrate with KFP Pipeline

For production, wrap the Dataflow job in a [KFP pipeline](https://www.kubeflow.org/docs/components/pipelines/) on Vertex AI Pipelines. This adds scheduling, parameterization, and lineage tracking.

The approach: write the Beam pipeline as a standalone Python script, upload it to GCS, then launch it from a KFP component using the Dataflow API.

### Standalone Pipeline Script

The `%%writefile` cell below creates a self-contained Beam pipeline script that can run independently or be launched by KFP. All parameters are passed via command-line arguments.

````python
%%writefile {DIR}/pipeline.py
"""Standalone Beam pipeline for sentiment batch inference."""
import subprocess, sys

# Fix package incompatibilities in the DataflowPythonJobOp launcher container
# (gcr.io/ml-pipeline/google-cloud-pipeline-components:2.22.0).
# The container has numpy/pandas ABI mismatches and broken apitools imports.
# Reinstall beam (with all deps) and pandas to get a consistent package set.
subprocess.check_call([
    sys.executable, '-m', 'pip', 'install', '-q',
    '--force-reinstall', 'apache-beam[gcp]', 'pandas',
])

import argparse, json, logging, os, tempfile

logging.basicConfig(level=logging.INFO)

import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions
from apache_beam.ml.inference.base import ModelHandler, PredictionResult, RunInference


class SentimentModelHandler(ModelHandler):
    def __init__(self, model_gcs_uri):
        self._model_gcs_uri = model_gcs_uri

    def load_model(self):
        from transformers import pipeline as hf_pipeline
        from google.cloud import storage
        client = storage.Client()
        parts = self._model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]
        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)
        return hf_pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)

    def run_inference(self, batch, model, inference_args=None):
        texts = [e['text'] for e in batch]
        results = model(texts)
        return [PredictionResult(example=e, inference=r) for e, r in zip(batch, results)]

    def batch_elements_kwargs(self):
        return {'min_batch_size': 1, 'max_batch_size': 32}


class PreprocessFn(beam.DoFn):
    def process(self, row):
        text = row.get('text', '')
        if isinstance(text, str):
            text = text.strip()
        if text:
            yield {'id': row.get('id', 0), 'text': text, 'category': row.get('category', '')}


class PostprocessFn(beam.DoFn):
    def process(self, prediction_result):
        element = prediction_result.example
        result = prediction_result.inference
        yield {
            'id': element['id'],
            'text': element['text'],
            'label': result['label'],
            'score': round(result['score'], 6),
            'model': 'distilbert',
        }


def run(argv=None):
    parser = argparse.ArgumentParser()
    # Note: --project is NOT declared here because DataflowPythonJobOp
    # auto-injects --project into the command. Declaring it here would
    # consume it via parse_known_args, preventing Beam from seeing it.
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--model_gcs_uri', required=True)
    known_args, pipeline_args = parser.parse_known_args(argv)

    # Get project from Beam pipeline options (auto-injected by the component)
    options = PipelineOptions(pipeline_args)
    project = options.get_all_options()['project']

    handler = SentimentModelHandler(known_args.model_gcs_uri)
    schema = 'id:INTEGER,text:STRING,label:STRING,score:FLOAT64,model:STRING'

    with beam.Pipeline(options=options) as p:
        (
            p
            | 'Read' >> beam.io.ReadFromBigQuery(
                query=f'SELECT id, text, category FROM `{project}.{known_args.bq_dataset}.{known_args.source_table}`',
                use_standard_sql=True,
            )
            | 'Preprocess' >> beam.ParDo(PreprocessFn())
            | 'Infer' >> RunInference(model_handler=handler)
            | 'Postprocess' >> beam.ParDo(PostprocessFn())
            | 'Write' >> beam.io.WriteToBigQuery(
                table=f'{project}:{known_args.bq_dataset}.{known_args.output_table}',
                schema=schema,
                write_disposition=beam.io.BigQueryDisposition.WRITE_TRUNCATE,
                create_disposition=beam.io.BigQueryDisposition.CREATE_IF_NEEDED,
            )
        )


if __name__ == '__main__':
    run()
````

### Define KFP Pipeline

Uses the pre-built [`DataflowPythonJobOp`](https://google-cloud-pipeline-components.readthedocs.io/en/google-cloud-pipeline-components-2.22.0/api/v1/dataflow.html) from `google-cloud-pipeline-components`. This component:

- Downloads the Python script and requirements from GCS
- Submits the Dataflow job
- Tracks the job via `gcp_resources` — Vertex AI Pipelines natively monitors the Dataflow job lifecycle (no active polling container)
- Automatically passes `--runner`, `--project`, `--region`, and `--temp_location` to the script
- Auto-stages GCS paths for `--requirements_file` to a local path before launching

````python
from kfp import dsl, compiler
from google_cloud_pipeline_components.v1.dataflow import DataflowPythonJobOp

@dsl.pipeline(
    name='dataflow-batch-inference',
    description='Run sentiment batch inference via Dataflow',
)
def dataflow_inference_pipeline(
    project: str,
    region: str,
    pipeline_gcs_path: str,
    requirements_gcs_path: str,
    model_gcs_uri: str,
    bq_dataset: str,
    source_table: str,
    output_table: str = 'pipeline_predictions',
    temp_location: str = '',
    staging_location: str = '',
    machine_type: str = 'n1-standard-4',
    num_workers: str = '2',
    max_workers: str = '4',
):
    # DataflowPythonJobOp automatically adds: --runner, --project, --region, --temp_location
    # It also auto-stages GCS paths for: --requirements_file, --setup_file
    # pipeline.py self-fixes the launcher container's numpy/pandas ABI mismatch.
    DataflowPythonJobOp(
        project=project,
        location=region,
        python_module_path=pipeline_gcs_path,
        temp_location=temp_location,
        args=[
            '--bq_dataset', bq_dataset,
            '--source_table', source_table,
            '--output_table', output_table,
            '--model_gcs_uri', model_gcs_uri,
            '--requirements_file', requirements_gcs_path,
            '--staging_location', staging_location,
            '--machine_type', machine_type,
            '--num_workers', num_workers,
            '--max_num_workers', max_workers,
            '--save_main_session',
        ],
    )

print('KFP pipeline defined: dataflow_inference_pipeline (using DataflowPythonJobOp)')
````

````python
# Compile and submit
pipeline_json = f'{DIR}/dataflow_inference_pipeline.json'
compiler.Compiler().compile(
    pipeline_func=dataflow_inference_pipeline,
    package_path=pipeline_json,
)
print(f'Pipeline compiled to {pipeline_json}')

pipeline_job = aiplatform.PipelineJob(
    display_name=f'{SERIES}-{EXPERIMENT}-pipeline',
    template_path=pipeline_json,
    parameter_values={
        'project': PROJECT_ID,
        'region': REGION,
        'pipeline_gcs_path': f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/dataflow/pipeline.py',
        'requirements_gcs_path': f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/dataflow/requirements.txt',
        'model_gcs_uri': MODELS['v1']['artifact_uri'],
        'bq_dataset': BQ_DATASET,
        'source_table': BQ_TABLE,
        'output_table': 'pipeline_predictions',
        'temp_location': f'gs://{GCS_BUCKET}/dataflow/temp',
        'staging_location': f'gs://{GCS_BUCKET}/dataflow/staging',
        'machine_type': 'n1-standard-4',
        'num_workers': '2',
        'max_workers': '4',
    },
    labels={'series': SERIES, 'experiment': EXPERIMENT},
    enable_caching=False,
)

print('Submitting pipeline ...')
pipeline_job.submit()
print(f'Pipeline job: {pipeline_job.resource_name}')
````

````python
# Wait for pipeline completion
pipeline_job.wait()
print(f'Pipeline state: {pipeline_job.state.name}')
````

## Configuration Reference

### Autoscaling

Dataflow autoscales workers between `num_workers` and `max_num_workers` based on backlog. This is a key advantage over Vertex AI Batch Prediction, which uses a fixed worker count.

| Parameter | Description | Default |
|-----------|-------------|---------|
| `num_workers` | Initial worker count | 1 |
| `max_num_workers` | Maximum workers (autoscaling ceiling) | 1000 |
| `machine_type` | Machine type per worker | `n1-standard-1` |
| `disk_size_gb` | Disk size per worker | 250 |

### Custom Containers

For production, pre-install large dependencies (torch, transformers) in a custom Docker image:

```python
options = PipelineOptions([
    '--sdk_container_image=us-central1-docker.pkg.dev/my-project/my-repo/beam-pytorch:latest',
    # No --requirements_file needed — deps are in the container
])
```

This reduces worker startup from ~5 minutes (installing torch) to ~30 seconds.

### GPU Support

Dataflow supports GPU workers via Runner v2:

```python
options = PipelineOptions([
    '--experiments=use_runner_v2',
    '--sdk_container_image=my-gpu-container',
    '--dataflow_service_options=worker_accelerator=type:nvidia-tesla-t4;count:1;install-nvidia-driver',
])
```

Not demonstrated here — our small sentiment models don't benefit from GPU acceleration.

### Comparison: Dataflow vs Vertex AI Batch Prediction

| Aspect | Dataflow RunInference | Vertex AI Batch Prediction |
|--------|:-:|:-:|
| **Pre/post processing** | Native pipeline stages | KFP pipeline steps |
| **Multi-model** | Single pipeline (branches) | Separate jobs |
| **Model loading** | Direct (ModelHandler) | Container-based |
| **Autoscaling** | Yes (within bounds) | No (fixed workers) |
| **Custom logic** | Full Beam SDK | Limited to container |
| **Dynamic batching** | BatchElements (auto-tuned) | Fixed batchSize |
| **Exactly-once** | Supported | At-least-once |
| **Best for** | Complex processing pipelines | Simple managed batch predict |
