# Batch Inference With Dataproc

Run batch inference using [PySpark](https://spark.apache.org/docs/latest/api/python/) on [Dataproc Serverless](https://cloud.google.com/dataproc-serverless/docs/overview). Models load directly into Spark executors — no container build, no endpoint. Dataproc auto-provisions resources, runs the job, and tears down.

This notebook demonstrates:

- Pandas UDF for HuggingFace model inference with per-executor model caching
- BigQuery I/O via the built-in Spark BigQuery connector
- Submitting batch jobs with the `BatchControllerClient` Python SDK
- Multi-model inference with separate batch jobs
- Orchestrating Dataproc jobs with Vertex AI Pipelines (KFP)

**How it works:** A PySpark script defines a Pandas UDF that loads a HuggingFace model from GCS once per executor, then processes text in batches. Dataproc Serverless runtime 2.2 ships with torch and transformers pre-installed — no custom container or dependency installation needed.

**Story arc:**
1. Write a PySpark inference script with a Pandas UDF
2. Submit to Dataproc Serverless and verify results
3. Compare two models (DistilBERT vs BERT) via separate batch jobs
4. Orchestrate the workflow as a KFP pipeline

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params) — returns `POSITIVE`/`NEGATIVE`
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params) — returns `LABEL_0`/`LABEL_1`

- **v2 (BERT):** Larger — second model for multi-model comparison

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

## Run on Dataproc Serverless

Deploy a PySpark batch job to [Dataproc Serverless](https://cloud.google.com/dataproc-serverless/docs/overview) for cloud-scale inference. The key features:

- **Serverless:** No cluster to manage — Dataproc provisions executors, runs the job, and tears down
- **Runtime 2.2:** Ships with torch, transformers, and other ML libs pre-installed — no custom container needed
- **BigQuery connector:** Built-in `spark.read.format('bigquery')` for direct BQ I/O
- **Autoscaling:** Dataproc adjusts executor count based on workload

The PySpark script uses a **Pandas UDF** pattern for inference:

```python
@pandas_udf(schema)
def sentiment_udf(texts: pd.Series) -> pd.DataFrame:
    pipe = get_model(model_uri)  # cached per executor
    results = pipe(texts.tolist(), batch_size=32)
    return pd.DataFrame(results)
```

Each executor loads the model from GCS once (cached in a module-level dict), then reuses it across all UDF calls. This avoids the per-row model loading overhead of a naive UDF.

````python
%%writefile {DIR}/inference.py
"""Standalone PySpark script for sentiment batch inference on Dataproc Serverless."""
import argparse
import os
import tempfile

from pyspark.sql import SparkSession
from pyspark.sql.functions import pandas_udf, col, lit, round as spark_round
from pyspark.sql.types import StructType, StructField, StringType, FloatType
import pandas as pd

# Module-level model cache — persists across UDF calls within the same executor
_model_cache = {}


def get_model(model_gcs_uri):
    """Load model from GCS, caching after first load per executor."""
    if model_gcs_uri not in _model_cache:
        from transformers import pipeline as hf_pipeline
        from google.cloud import storage

        client = storage.Client()
        parts = model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]
        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)
        # Load tokenizer explicitly — DistilBERT doesn't accept token_type_ids
        # but some transformers versions (e.g. runtime 2.2) generate them
        from transformers import AutoTokenizer
        tokenizer = AutoTokenizer.from_pretrained(model_dir)
        if 'token_type_ids' in tokenizer.model_input_names:
            tokenizer.model_input_names.remove('token_type_ids')
        _model_cache[model_gcs_uri] = hf_pipeline(
            'sentiment-analysis', model=model_dir, tokenizer=tokenizer
        )
    return _model_cache[model_gcs_uri]


def run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--project', required=True)
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--model_gcs_uri', required=True)
    parser.add_argument('--model_name', default='distilbert')
    parser.add_argument('--temp_bucket', required=True)
    args = parser.parse_args()

    spark = SparkSession.builder.appName('batch-inference').getOrCreate()
    spark.conf.set('temporaryGcsBucket', args.temp_bucket)

    # Read from BigQuery
    df = spark.read.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.source_table}') \
        .load() \
        .select('id', 'text', 'category')

    # Define sentiment UDF with struct return type
    model_uri = args.model_gcs_uri

    schema = StructType([
        StructField('label', StringType()),
        StructField('score', FloatType()),
    ])

    @pandas_udf(schema)
    def sentiment_udf(texts: pd.Series) -> pd.DataFrame:
        pipe = get_model(model_uri)
        results = pipe(texts.tolist(), batch_size=32)
        return pd.DataFrame(results)

    # Apply inference and format output
    result_df = df.withColumn('prediction', sentiment_udf(col('text')))
    output_df = result_df.select(
        col('id'),
        col('text'),
        col('prediction.label').alias('label'),
        spark_round(col('prediction.score'), 6).alias('score'),
        lit(args.model_name).alias('model'),
    )

    # Write to BigQuery
    output_df.write.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.output_table}') \
        .mode('overwrite') \
        .save()

    print(f'Results written to {args.project}.{args.bq_dataset}.{args.output_table}')


if __name__ == '__main__':
    run()
````

````python
# Upload PySpark script to GCS
bucket.blob(f'{SERIES}/{EXPERIMENT}/dataproc/inference.py').upload_from_filename(
    f'{DIR}/inference.py'
)
script_gcs_path = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/dataproc/inference.py'
print(f'Script uploaded: {script_gcs_path}')
````

````python
# Submit Dataproc Serverless batch job
BQ_OUTPUT_TABLE = 'dataproc_predictions'
batch_id = f'sentiment-batch-{datetime.now().strftime("%Y%m%d-%H%M%S")}'

batch = dataproc_v1.Batch(
    pyspark_batch=dataproc_v1.PySparkBatch(
        main_python_file_uri=script_gcs_path,
        args=[
            '--project', PROJECT_ID,
            '--bq_dataset', BQ_DATASET,
            '--source_table', BQ_TABLE,
            '--output_table', BQ_OUTPUT_TABLE,
            '--model_gcs_uri', MODELS['v1']['artifact_uri'],
            '--model_name', 'distilbert',
            '--temp_bucket', GCS_BUCKET,
        ],
    ),
    runtime_config=dataproc_v1.RuntimeConfig(version='2.2'),
)

print(f'Submitting Dataproc batch job: {batch_id}')
operation = dataproc_client.create_batch(
    request={
        'parent': f'projects/{PROJECT_ID}/locations/{REGION}',
        'batch': batch,
        'batch_id': batch_id,
    }
)
print(f'Job submitted. Console: https://console.cloud.google.com/dataproc/batches/{REGION}/{batch_id}?project={PROJECT_ID}')
````

````python
# Wait for job completion (blocks until SUCCEEDED/FAILED/CANCELLED)
response = operation.result()
print(f'Batch state: {dataproc_v1.Batch.State(response.state).name}')
````

### Inference Patterns

| Pattern | Model Loading | Best For |
|---------|-------------|----------|
| **Pandas UDF** (shown above) | Once per executor (cached in module dict) | Structured DataFrames, SQL-friendly output |
| **`mapPartitions`** | Once per partition (explicit) | Full control over model lifecycle, RDD-based |
| **Naive UDF** | Every row (slow!) | Never use for ML inference |

The **Pandas UDF** pattern is preferred because:
- Processes data in Arrow-optimized batches (not row-by-row)
- Returns structured types (struct with `label` + `score`)
- Integrates with DataFrame operations (`withColumn`, `select`)
- Model caching via module-level dict avoids repeated downloads

**`mapPartitions` alternative** (lower-level):

```python
def predict_partition(iterator):
    pipe = get_model(model_uri)  # loaded once per partition
    for row in iterator:
        result = pipe(row.text)[0]
        yield (row.id, row.text, result['label'], result['score'], 'distilbert')

results_rdd = df.rdd.mapPartitions(predict_partition)
output_df = spark.createDataFrame(results_rdd, schema=['id', 'text', 'label', 'score', 'model'])
```

## Multi-Model Inference

Run both DistilBERT and BERT on the same data via separate Dataproc Serverless batch jobs. Each job writes to a different output table, then we compare results.

Unlike Dataflow (which can branch within a single pipeline), Dataproc Serverless runs one PySpark script per batch job. Sequential jobs are the simplest pattern — each job is independent and uses the same script with different `--model_gcs_uri` and `--output_table` args.

````python
# Submit batch job for v2 (BERT) model
BQ_OUTPUT_TABLE_V2 = 'dataproc_predictions_v2'
batch_id_v2 = f'sentiment-batch-v2-{datetime.now().strftime("%Y%m%d-%H%M%S")}'

batch_v2 = dataproc_v1.Batch(
    pyspark_batch=dataproc_v1.PySparkBatch(
        main_python_file_uri=script_gcs_path,
        args=[
            '--project', PROJECT_ID,
            '--bq_dataset', BQ_DATASET,
            '--source_table', BQ_TABLE,
            '--output_table', BQ_OUTPUT_TABLE_V2,
            '--model_gcs_uri', MODELS['v2']['artifact_uri'],
            '--model_name', 'bert',
            '--temp_bucket', GCS_BUCKET,
        ],
    ),
    runtime_config=dataproc_v1.RuntimeConfig(version='2.2'),
)

print(f'Submitting v2 batch job: {batch_id_v2}')
operation_v2 = dataproc_client.create_batch(
    request={
        'parent': f'projects/{PROJECT_ID}/locations/{REGION}',
        'batch': batch_v2,
        'batch_id': batch_id_v2,
    }
)
print(f'Job submitted. Console: https://console.cloud.google.com/dataproc/batches/{REGION}/{batch_id_v2}?project={PROJECT_ID}')
````

````python
# Wait for v2 job
response_v2 = operation_v2.result()
print(f'Batch state: {dataproc_v1.Batch.State(response_v2.state).name}')
````

## Orchestrate with KFP Pipeline

For production, wrap the Dataproc job in a [KFP pipeline](https://www.kubeflow.org/docs/components/pipelines/) on Vertex AI Pipelines. This adds scheduling, parameterization, and lineage tracking.

The approach: the PySpark script is already uploaded to GCS. The KFP pipeline uses the pre-built `DataprocPySparkBatchOp` to submit and monitor the job.

### Define KFP Pipeline

Uses the pre-built [`DataprocPySparkBatchOp`](https://google-cloud-pipeline-components.readthedocs.io/en/google-cloud-pipeline-components-2.22.0/api/v1/dataproc.html) from `google-cloud-pipeline-components`. This component:

- Submits a Dataproc Serverless batch job
- Tracks the job via `gcp_resources` — Vertex AI Pipelines natively monitors the batch lifecycle (no active polling container)
- Supports runtime version, Spark properties, custom containers, and all batch parameters

````python
from kfp import dsl, compiler
from google_cloud_pipeline_components.v1.dataproc import DataprocPySparkBatchOp

@dsl.pipeline(
    name='dataproc-batch-inference',
    description='Run sentiment batch inference via Dataproc Serverless',
)
def dataproc_inference_pipeline(
    project: str,
    region: str,
    script_gcs_path: str,
    model_gcs_uri: str,
    model_name: str,
    bq_dataset: str,
    source_table: str,
    output_table: str = 'pipeline_predictions',
    temp_bucket: str = '',
):
    DataprocPySparkBatchOp(
        project=project,
        location=region,
        main_python_file_uri=script_gcs_path,
        runtime_config_version='2.2',
        runtime_config_properties={
            'spark.executor.instances': '2',
            'spark.driver.memory': '4g',
            'spark.executor.memory': '4g',
        },
        args=[
            '--project', project,
            '--bq_dataset', bq_dataset,
            '--source_table', source_table,
            '--output_table', output_table,
            '--model_gcs_uri', model_gcs_uri,
            '--model_name', model_name,
            '--temp_bucket', temp_bucket,
        ],
    )

print('KFP pipeline defined: dataproc_inference_pipeline (using DataprocPySparkBatchOp)')
````

````python
# Compile and submit
pipeline_json = f'{DIR}/dataproc_inference_pipeline.json'
compiler.Compiler().compile(
    pipeline_func=dataproc_inference_pipeline,
    package_path=pipeline_json,
)
print(f'Pipeline compiled to {pipeline_json}')

pipeline_job = aiplatform.PipelineJob(
    display_name=f'{SERIES}-{EXPERIMENT}-pipeline',
    template_path=pipeline_json,
    parameter_values={
        'project': PROJECT_ID,
        'region': REGION,
        'script_gcs_path': f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/dataproc/inference.py',
        'model_gcs_uri': MODELS['v1']['artifact_uri'],
        'model_name': 'distilbert',
        'bq_dataset': BQ_DATASET,
        'source_table': BQ_TABLE,
        'output_table': 'pipeline_predictions',
        'temp_bucket': GCS_BUCKET,
    },
    labels={'series': SERIES, 'experiment': EXPERIMENT},
    enable_caching=False,
)

print('Submitting pipeline ...')
pipeline_job.submit()
print(f'Pipeline job: {pipeline_job.resource_name}')
````

````python
# Wait for pipeline completion
pipeline_job.wait()
print(f'Pipeline state: {pipeline_job.state.name}')
````

````python
# Read KFP pipeline output
kfp_table = f'{PROJECT_ID}.{BQ_DATASET}.pipeline_predictions'
try:
    kfp_results = bq_client.query(f'SELECT * FROM `{kfp_table}` ORDER BY id').to_dataframe()
    print(f'Pipeline results ({len(kfp_results)} rows):')
except Exception as e:
    kfp_results = None
    print(f'Could not read results table: {e}')

kfp_results
````

## Configuration Reference

### Dataproc Serverless

Dataproc Serverless auto-provisions resources per job. Key configuration via Spark properties:

| Property | Description | Default |
|----------|-------------|----------|
| `spark.executor.instances` | Number of executors | 2 |
| `spark.executor.memory` | Memory per executor | 1g |
| `spark.executor.cores` | Cores per executor | 2 |
| `spark.driver.memory` | Driver memory | 1g |
| `spark.dynamicAllocation.enabled` | Enable autoscaling | true |

### Runtime Versions

| Version | Key ML Libraries |
|---------|------------------|
| **2.2** | torch, transformers, tokenizers, accelerate, xgboost |
| 2.1 | Standard Spark + BigQuery connector |
| 1.x | Legacy — no ML libs pre-installed |

Use runtime 2.2+ for ML workloads to avoid dependency installation.

### Custom Containers

For full control over the environment (pinned versions, additional native libraries):

```python
batch.runtime_config = dataproc_v1.RuntimeConfig(
    container_image='us-central1-docker.pkg.dev/project/repo/image:tag',
)
```

The container must NOT include Spark (it's mounted at runtime). See [custom container docs](https://cloud.google.com/dataproc-serverless/docs/guides/custom-containers).

### GPU Support

Dataproc supports GPU via RAPIDS accelerator — not demonstrated here as our small models don't benefit from GPU acceleration.

### Three-Way Comparison

| Aspect | Vertex AI Batch | Dataflow RunInference | Dataproc PySpark |
|--------|:-:|:-:|:-:|
| **Processing model** | Container per worker | Beam pipeline | Spark DataFrame |
| **Pre/post processing** | KFP pipeline steps | Beam DoFns | Spark UDFs / transforms |
| **Model loading** | Container startup | ModelHandler | Pandas UDF + cache |
| **Autoscaling** | No (fixed workers) | Yes (within bounds) | Yes (dynamic allocation) |
| **Data ecosystem** | GCS, BQ | GCS, BQ, Pub/Sub | GCS, BQ, Hive, Delta |
| **Multi-model** | Separate jobs | Single pipeline branches | Separate jobs |
| **Serverless** | Yes | Yes | Yes |
| **ML libs pre-installed** | No (container) | No (requirements) | Yes (runtime 2.2) |
| **Best for** | Simple managed predict | Complex processing | Spark-native teams |
