# Vertex AI Batch Prediction

Run batch inference on large datasets using [Vertex AI Batch Prediction](https://cloud.google.com/vertex-ai/docs/predictions/get-batch-predictions). This notebook demonstrates:

- Batch prediction from GCS (JSONL) and BigQuery inputs
- Column control with `instanceConfig`: `includedFields`, `excludedFields`
- Tuning worker count and batch size for optimal throughput
- Multi-model batch prediction for comparison and validation
- Orchestrating batch workflows with Vertex AI Pipelines (KFP)

**Why batch?** When you need predictions for thousands or millions of records and don't need real-time responses, batch prediction is more cost-effective than online endpoints. No persistent infrastructure, no idle compute, and native integration with BigQuery for warehouse-scale scoring.

**Container portability:** The same FastAPI container used in the [online serving notebooks](../readme.md) works for batch prediction. Vertex AI manages the workers, distributes input data, collects results, and shuts down. No endpoint needed.

**Story arc:**
1. Build container and register models
2. Batch prediction from GCS (JSONL) with DistilBERT
3. Batch prediction from BigQuery with BERT + `instanceConfig` column control
4. Compare DistilBERT vs BERT results (multi-model)
5. Orchestrate the full workflow as a KFP pipeline

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params) — returns `POSITIVE`/`NEGATIVE`
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params) — returns `LABEL_0`/`LABEL_1`

## Download HuggingFace Models

Download two sentiment analysis models from HuggingFace and upload them to GCS. Both models are fine-tuned on the [SST-2](https://huggingface.co/datasets/stanfordnlp/sst2) dataset for binary sentiment classification.

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)
for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']

    # Handle both plain strings (GCS JSONL) and objects (BQ / instanceConfig)
    texts = []
    for inst in instances:
        if isinstance(inst, str):
            texts.append(inst)
        elif isinstance(inst, dict):
            # instanceConfig sends filtered objects — extract the text field
            texts.append(inst.get('text', str(inst)))
        else:
            texts.append(str(inst))

    results = classifier(texts)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
````

## Batch Prediction from GCS (JSONL)

Run batch prediction using JSONL input from GCS. Each line in the input file is a JSON string — the model receives these as plain text instances.

This uses the high-level `model.batch_predict()` SDK which blocks until the job completes. We configure the worker count and machine type to control throughput and cost.

````python
# Create JSONL input file — each line is a JSON-encoded string
jsonl_content = '\n'.join(json.dumps(s) for s in SENTENCES)
jsonl_blob = bucket.blob(f'{SERIES}/{EXPERIMENT}/batch/input/sentences.jsonl')
jsonl_blob.upload_from_string(jsonl_content)

jsonl_gcs_uri = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/batch/input/sentences.jsonl'
print(f'Input file: {jsonl_gcs_uri}')
print(f'Sentences: {len(SENTENCES)}')
print()
print('Preview:')
for line in jsonl_content.split('\n')[:3]:
    print(f'  {line}')
````

````python
print('Submitting batch prediction job (DistilBERT, GCS JSONL) ...')
print(f'  Workers: 2 x n1-standard-4')
print()

batch_job_gcs = vertex_model_v1.batch_predict(
    job_display_name=f'{SERIES}-{EXPERIMENT}-gcs-distilbert',
    gcs_source=[jsonl_gcs_uri],
    gcs_destination_prefix=f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/batch/output/gcs',
    instances_format='jsonl',
    machine_type='n1-standard-4',
    starting_replica_count=2,
    max_replica_count=2,
)

print(f'Job completed: {batch_job_gcs.resource_name}')
print(f'Duration: {batch_job_gcs.end_time - batch_job_gcs.start_time}')
````

````python
# Read predictions from GCS output
output_dir = batch_job_gcs.output_info.gcs_output_directory
prefix = output_dir.replace(f'gs://{GCS_BUCKET}/', '')

gcs_results = []
for blob in bucket.list_blobs(prefix=prefix):
    # Vertex AI writes output files named like prediction.results-00000-of-00001 (JSONL format, no .jsonl extension)
    if 'prediction' in blob.name.split('/')[-1]:
        for line in blob.download_as_text().strip().split('\n'):
            if line:
                gcs_results.append(json.loads(line))

print(f'DistilBERT predictions ({len(gcs_results)} sentences):')
print()
for r in gcs_results:
    text = r['instance'] if isinstance(r['instance'], str) else str(r['instance'])
    pred = r['prediction']
    print(f'  "{text[:60]}..."')
    print(f'    => {pred["label"]} (score: {pred["score"]:.6f})')
    print()
````

### Worker Configuration

Batch prediction workers are configured at job creation:

| Parameter | Description | Default |
|-----------|-------------|---------|
| `starting_replica_count` | Number of workers to spin up. No autoscaling in batch mode — this is the actual count. | 1 |
| `max_replica_count` | Should equal `starting_replica_count` for batch (no autoscaling). | 1 |
| `machine_type` | Machine type for each worker. Larger machines = more memory for bigger models. | `n1-standard-2` |

**Batch size tuning:** Vertex AI also supports `manualBatchTuningParameters.batchSize` (via the REST/gapic API) which controls how many instances are sent per HTTP request to the container. The default is 64. Too large = OOM in the container; too small = underutilized workers. See the [Tuning Reference](#tuning--monitoring-reference) section.

**Cost:** Workers run for the duration of the job. Cost = worker_count x machine_type_rate x job_duration. More workers = faster but more expensive — linear scaling for embarrassingly parallel workloads.

## Batch Prediction from BigQuery with `instanceConfig`

Run batch prediction using a BigQuery table as input, with `instanceConfig` to control which columns the model sees. This is the production pattern: your BQ table has many columns but the model only needs one.

`instanceConfig` provides three controls:
- **`included_fields`** — Whitelist of columns to send to the model. Only these fields become the instance.
- **`excluded_fields`** — Blacklist of columns to exclude (mutually exclusive with `included_fields`).
- **`key_field`** — A column that passes through to the output but is **not** sent to the model. Only supported with GCS output format — not with BigQuery output.

We include `id` and `text` in `included_fields` so `id` is available in the output for joining back to the source table. The model container handles the extra field gracefully — it extracts the `text` field from the instance dict.

This section uses the lower-level `aiplatform_v1` gapic API because the high-level `model.batch_predict()` SDK does not expose `instanceConfig`. We use BERT (v2) here so we can compare with DistilBERT (v1) from the GCS section above.

````python
# Create BQ dataset
dataset_ref = bigquery.Dataset(f'{PROJECT_ID}.{BQ_DATASET}')
dataset_ref.location = REGION
try:
    bq_client.get_dataset(dataset_ref)
    print(f'Dataset {BQ_DATASET} already exists')
except Exception:
    bq_client.create_dataset(dataset_ref)
    print(f'Created dataset {BQ_DATASET}')

# Create input table with multiple columns (to demonstrate instanceConfig filtering)
BQ_TABLE = 'input_data'
schema = [
    bigquery.SchemaField('id', 'INTEGER'),
    bigquery.SchemaField('text', 'STRING'),
    bigquery.SchemaField('category', 'STRING'),
    bigquery.SchemaField('created_date', 'DATE'),
]

# Build rows
categories = ['entertainment', 'dining', 'general', 'product', 'product',
              'weather', 'service', 'entertainment', 'product', 'general']
rows = [
    {'id': i + 1, 'text': text, 'category': categories[i], 'created_date': '2026-04-18'}
    for i, text in enumerate(SENTENCES)
]

# Load with WRITE_TRUNCATE to replace any existing data (idempotent across re-runs)
job_config = bigquery.LoadJobConfig(
    schema=schema,
    write_disposition='WRITE_TRUNCATE',
)
bq_client.load_table_from_json(
    rows, f'{PROJECT_ID}.{BQ_DATASET}.{BQ_TABLE}', job_config=job_config
).result()
print(f'Loaded {len(rows)} rows into {BQ_TABLE} (WRITE_TRUNCATE)')

# Preview
bq_client.query(f'SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.{BQ_TABLE}` ORDER BY id LIMIT 3').to_dataframe()
````

````python
# Build batch prediction job with instanceConfig using gapic API
batch_prediction_job = aiplatform_v1.BatchPredictionJob(
    display_name=f'{SERIES}-{EXPERIMENT}-bq-bert-instanceconfig',
    model=vertex_model_v2.resource_name,
    input_config={
        'instances_format': 'bigquery',
        'bigquery_source': {
            'input_uri': f'bq://{PROJECT_ID}.{BQ_DATASET}.{BQ_TABLE}',
        },
    },
    output_config={
        'predictions_format': 'bigquery',
        'bigquery_destination': {
            'output_uri': f'bq://{PROJECT_ID}.{BQ_DATASET}',
        },
    },
    dedicated_resources={
        'machine_spec': {'machine_type': 'n1-standard-4'},
        'starting_replica_count': 2,
        'max_replica_count': 2,
    },
    # instanceConfig: control which columns the model sees
    # Note: key_field is only supported with GCS output, not BigQuery output.
    # Include 'id' in included_fields so it appears in the output for joining.
    instance_config={
        'included_fields': ['id', 'text'],  # only send these columns to the model
        # 'excluded_fields': ['category', 'created_date'],  # alternative approach
    },
)

print('Submitting batch prediction job (BERT, BQ + instanceConfig) ...')
print(f'  included_fields: [id, text] (only columns sent to model)')
print()

response = jobs_client.create_batch_prediction_job(
    parent=f'projects/{PROJECT_ID}/locations/{REGION}',
    batch_prediction_job=batch_prediction_job,
)
print(f'Job created: {response.name}')
````

### `instanceConfig` Reference

| Field | Purpose | Output Format | Example |
|-------|---------|:---:|---------|
| `included_fields` | Whitelist: only these columns are sent as the instance. | GCS, BQ | `['id', 'text']` |
| `excluded_fields` | Blacklist: these columns are excluded from the instance. Mutually exclusive with `included_fields`. | GCS, BQ | `['category', 'created_date']` |
| `key_field` | Column that passes through to output without being sent to the model. Use as a join key. | GCS only | `'id'` |
| `instance_type` | How to format the instance: `'object'` (default for structured), `'array'` (ordered values). | GCS, BQ | `'object'` |

> **`key_field` limitation:** `key_field` is only supported when the output format is GCS (JSONL). With BigQuery output, use `included_fields` to include your ID column — it flows through to the output table naturally. The model container should handle the extra field gracefully (as our container does by extracting only the `text` field).

**Why use `instanceConfig`?**

- **Efficiency:** Your BQ table might have 50 columns but the model only needs 1. Without `instanceConfig`, all 50 columns are serialized and sent to the model container — wasted bandwidth and potential errors.
- **Column filtering:** `included_fields` / `excluded_fields` let you control exactly what the model sees, without creating a view or staging table.
- **No data prep needed:** Configure column filtering directly in the batch job definition.

> **Note:** `instanceConfig` requires the gapic API (`aiplatform_v1.JobServiceClient`). The high-level `model.batch_predict()` SDK does not expose this parameter. For simple batch jobs without column filtering, `model.batch_predict()` is sufficient — as shown in the GCS section above.

## Pre/Post Processing with KFP Pipeline

Batch prediction rarely stands alone in production. Typically there's a workflow:
1. **Pre-process:** Clean and filter input data (SQL query, feature engineering)
2. **Predict:** Run batch prediction on the prepared data
3. **Post-process:** Transform predictions (map labels, compute confidence buckets, join to source)

[Kubeflow Pipelines (KFP)](https://www.kubeflow.org/docs/components/pipelines/) on Vertex AI Pipelines orchestrates this as a three-stage pipeline. Each stage runs in its own container, passes outputs to the next, and the full workflow is tracked with lineage and artifacts.

### Define Pipeline Components

Three custom KFP components, each installing only the packages it needs:
- **`preprocess_bq`** — Runs a BQ query to clean and stage data
- **`batch_predict_op`** — Calls `model.batch_predict()` on the staged data
- **`postprocess_bq`** — Transforms predictions with a BQ query (label mapping, confidence buckets, join to source)

````python
from kfp import dsl, compiler

@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-bigquery']
)
def preprocess_bq(
    project: str,
    bq_dataset: str,
    source_table: str,
    staging_table: str,
) -> str:
    """Clean and filter data in BigQuery, write to staging table."""
    from google.cloud import bigquery
    client = bigquery.Client(project=project)

    query = f"""
    CREATE OR REPLACE TABLE `{project}.{bq_dataset}.{staging_table}` AS
    SELECT DISTINCT
        id,
        TRIM(text) AS text
    FROM `{project}.{bq_dataset}.{source_table}`
    WHERE text IS NOT NULL
        AND LENGTH(TRIM(text)) > 0
    ORDER BY id
    """
    client.query(query).result()
    count = client.query(f'SELECT COUNT(*) c FROM `{project}.{bq_dataset}.{staging_table}`').result()
    for row in count:
        print(f'Staging table {staging_table}: {row.c} rows')
    return f'bq://{project}.{bq_dataset}.{staging_table}'


@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-aiplatform']
)
def batch_predict_op(
    project: str,
    location: str,
    model_resource_name: str,
    bigquery_source: str,
    bigquery_destination_prefix: str,
    bq_dataset: str,
    machine_type: str,
    replica_count: int,
) -> str:
    """Run Vertex AI batch prediction on a BQ table."""
    from google.cloud import aiplatform
    aiplatform.init(project=project, location=location)

    model = aiplatform.Model(model_resource_name)
    job = model.batch_predict(
        job_display_name='pipeline-batch-predict',
        bigquery_source=bigquery_source,
        bigquery_destination_prefix=bigquery_destination_prefix,
        instances_format='bigquery',
        machine_type=machine_type,
        starting_replica_count=replica_count,
        max_replica_count=replica_count,
    )
    output_table = job.output_info.bigquery_output_table
    # Normalize table reference — may be just the table name without project.dataset
    output_table = output_table.replace('bq://', '').replace(':', '.')
    if output_table.count('.') < 2:
        output_table = f'{project}.{bq_dataset}.{output_table}'
    print(f'Batch prediction output: {output_table}')
    return output_table


@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-bigquery']
)
def postprocess_bq(
    project: str,
    bq_dataset: str,
    predictions_table: str,
    source_table: str,
    output_table: str,
) -> str:
    """Transform predictions: parse JSON, join to source for enrichment."""
    from google.cloud import bigquery
    client = bigquery.Client(project=project)

    # The predictions table has: id, text, prediction (JSON string)
    # Post-processing: parse prediction JSON, join to source for category
    query = f"""
    CREATE OR REPLACE TABLE `{project}.{bq_dataset}.{output_table}` AS
    SELECT DISTINCT
        p.id,
        p.text,
        JSON_VALUE(p.prediction, '$.label') AS predicted_label,
        CAST(JSON_VALUE(p.prediction, '$.score') AS FLOAT64) AS confidence,
        s.category,
        CURRENT_TIMESTAMP() AS scored_at
    FROM `{predictions_table}` p
    LEFT JOIN (SELECT DISTINCT id, category FROM `{project}.{bq_dataset}.{source_table}`) s USING (id)
    ORDER BY p.id
    """
    client.query(query).result()
    count = client.query(f'SELECT COUNT(*) c FROM `{project}.{bq_dataset}.{output_table}`').result()
    for row in count:
        print(f'Results table {output_table}: {row.c} rows')
    return f'{project}.{bq_dataset}.{output_table}'

print('KFP components defined: preprocess_bq, batch_predict_op, postprocess_bq')
````

````python
@dsl.pipeline(
    name='batch-prediction-pipeline',
    description='Pre-process data, run batch prediction, post-process results',
)
def batch_prediction_pipeline(
    project: str,
    location: str,
    model_resource_name: str,
    bq_dataset: str,
    source_table: str,
    machine_type: str = 'n1-standard-4',
    replica_count: int = 2,
):
    # Stage 1: Clean and filter data
    preprocess_task = preprocess_bq(
        project=project,
        bq_dataset=bq_dataset,
        source_table=source_table,
        staging_table=f'{source_table}_staging',
    )

    # Stage 2: Run batch prediction on staged data
    predict_task = batch_predict_op(
        project=project,
        location=location,
        model_resource_name=model_resource_name,
        bigquery_source=preprocess_task.output,
        bigquery_destination_prefix=f'bq://{project}.{bq_dataset}',
        bq_dataset=bq_dataset,
        machine_type=machine_type,
        replica_count=replica_count,
    )

    # Stage 3: Transform predictions — join back to source, add metadata
    postprocess_task = postprocess_bq(
        project=project,
        bq_dataset=bq_dataset,
        predictions_table=predict_task.output,
        source_table=source_table,
        output_table=f'{source_table}_results',
    )

print('Pipeline defined: batch_prediction_pipeline')
````

````python
# Compile the pipeline to JSON
pipeline_json = f'{DIR}/batch_prediction_pipeline.json'
compiler.Compiler().compile(
    pipeline_func=batch_prediction_pipeline,
    package_path=pipeline_json,
)
print(f'Pipeline compiled to {pipeline_json}')

# Submit to Vertex AI Pipelines
pipeline_job = aiplatform.PipelineJob(
    display_name=f'{SERIES}-{EXPERIMENT}-pipeline',
    template_path=pipeline_json,
    parameter_values={
        'project': PROJECT_ID,
        'location': REGION,
        'model_resource_name': vertex_model_v1.resource_name,
        'bq_dataset': BQ_DATASET,
        'source_table': BQ_TABLE,
        'machine_type': 'n1-standard-4',
        'replica_count': 2,
    },
    labels={'series': SERIES, 'experiment': EXPERIMENT},
    enable_caching=False,
)

print('Submitting pipeline ...')
pipeline_job.submit()
print(f'Pipeline job: {pipeline_job.resource_name}')
````

````python
# Wait for pipeline completion
pipeline_job.wait()
print(f'Pipeline state: {pipeline_job.state.name}')
````

````python
# Read the final output table
results_table = f'{PROJECT_ID}.{BQ_DATASET}.{BQ_TABLE}_results'
try:
    pipeline_results = bq_client.query(f'SELECT * FROM `{results_table}` ORDER BY id').to_dataframe()
    print(f'Pipeline results ({len(pipeline_results)} rows):')
except Exception as e:
    pipeline_results = None
    print(f'Could not read results table: {e}')
    print('Check the pipeline run in the Vertex AI console for details.')

pipeline_results
````

## Tuning & Monitoring Reference

### Batch Size

The `batchSize` parameter (via `manualBatchTuningParameters` in the REST/gapic API) controls how many instances are sent per HTTP request to the container:

| Batch Size | Effect |
|:---:|--------|
| **Too large** | Container OOMs, failed requests, automatic retries slow the job |
| **Too small** | Underutilized GPU/CPU, more HTTP overhead, slower throughput |
| **Default (64)** | Good starting point for most models |

**Rule of thumb:** Start with the default (64). If the container logs show OOM errors, reduce by half. If container CPU/memory utilization is low, increase. For transformer models like ours, 16-32 is often optimal due to the memory footprint per inference.

### Worker Count

| Factor | Guidance |
|--------|----------|
| **Dataset size** | 10K instances: 1-2 workers. 1M instances: 10-20 workers. |
| **Model complexity** | Large models need more memory per worker → use larger machines, not more workers |
| **Cost vs speed** | Linear scaling: 2x workers ≈ 2x cost, ~0.5x duration |
| **Startup overhead** | Each worker pulls the container image + downloads model from GCS. For small datasets, 1-2 workers avoid spending more time on startup than inference. |

### Monitoring

- **Job state:** Poll via SDK (`job.state`) or check the [Batch Predictions console](https://console.cloud.google.com/vertex-ai/batch-predictions)
- **Container logs:** Available in Cloud Logging under the batch prediction job resource
- **Errors:** Check `job.partial_failures` for per-instance errors (malformed input, model errors)

### Cost

Batch prediction cost = `worker_count` x `machine_type_rate` x `job_duration`. There is no charge for idle time — workers run only for the duration of the job. For cost comparison with online endpoints, batch is typically 3-10x cheaper for large datasets because there's no persistent infrastructure.
