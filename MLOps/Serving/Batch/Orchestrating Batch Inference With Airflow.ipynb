# Orchestrating Batch Inference With Airflow

Orchestrate batch inference workflows using [Cloud Composer](https://cloud.google.com/composer/docs/concepts/overview) (managed [Apache Airflow](https://airflow.apache.org/)). Airflow adds what standalone batch jobs lack: scheduling, data dependencies, monitoring, alerting, and backfill.

This notebook demonstrates four orchestration patterns:

1. **Airflow → Dataproc** — `DataprocCreateBatchOperator` launches a PySpark Serverless job directly
2. **Airflow → Dataflow** — `BeamRunPythonPipelineOperator` launches a Beam pipeline on Dataflow
3. **Airflow → KFP Pipeline** — `RunPipelineJobOperator` triggers a Vertex AI Pipeline that runs inference
4. **Airflow → Vertex AI Batch** — `CreateBatchPredictionJobOperator` for managed batch prediction

**Where Airflow fits:** KFP (Notebooks 1-3) handles the *how* — ML pipeline logic, artifact tracking, model lineage. Airflow handles the *when* — scheduling, data dependencies, cross-system coordination, backfill. In production, Airflow is the scheduler that decides when to kick off the ML pipeline.

**Prerequisites:** This notebook is fully self-contained — it builds the prediction container and registers the model for Pattern 4 inline.

## Download HuggingFace Models

Download two sentiment analysis models from HuggingFace and upload them to GCS. Dataproc workers, Dataflow workers, and the Vertex AI Batch Prediction container all download these model files at job startup.

- **v1 (DistilBERT):** Smaller, faster — primary model
- **v2 (BERT):** Larger — second model for multi-model comparison


````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

````python
for version, info in MODELS.items():
    blobs = list(bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/models/{version}'))
    print(f'{version} ({info["display_name"]}): {len(blobs)} files at {info["artifact_uri"]}')
````

## Why Airflow for Batch ML

KFP (Notebooks 1-3) handles **how** to run inference — pipeline logic, artifact tracking, model lineage. Airflow handles **when** and **what around it**:

**Scheduling:** Cron-based or event-driven triggers. Run inference daily at 2 AM, or when new data lands.

**Data Dependencies:** `BigQueryCheckOperator` waits for today's partition to exist before starting inference — no wasted compute on missing data.

**Backfill:** Missed a day? `airflow dags backfill --start-date 2026-04-15 --end-date 2026-04-17` reruns the DAG for each missing date. KFP has no native equivalent.

**Cross-System Coordination:** Airflow DAGs can combine BQ checks, ML inference, Slack notifications, and downstream ETL in one workflow.

**Monitoring & Alerting:** Built-in retry, SLA tracking, failure callbacks. Page the on-call when inference fails or misses its deadline.


### Production Flow

```
Data Landing (scheduled/event)
    |
    v
Data Validation (Airflow: BigQueryCheckOperator)
    |
    v
ML Inference (Airflow triggers: Dataproc / Dataflow / Vertex AI / KFP)
    |
    v
Result Validation (Airflow: BigQueryCheckOperator on output)
    |
    v
Downstream Systems (Airflow: BQ transforms, notifications, exports)
```

## Prepare Inference Scripts

The Dataproc and Dataflow DAGs launch standalone Python scripts. These are the same proven scripts from [Notebook 2 (Dataflow)](./Batch%20Inference%20With%20Dataflow.ipynb) and [Notebook 3 (Dataproc)](./Batch%20Inference%20With%20Dataproc.ipynb) — copied here so this notebook is self-contained.


````python
%%writefile {DIR}/inference.py
"""Standalone PySpark script for sentiment batch inference on Dataproc Serverless."""
import argparse
import os
import tempfile

from pyspark.sql import SparkSession
from pyspark.sql.functions import pandas_udf, col, lit, round as spark_round
from pyspark.sql.types import StructType, StructField, StringType, FloatType
import pandas as pd

# Module-level model cache — persists across UDF calls within the same executor
_model_cache = {}


def get_model(model_gcs_uri):
    """Load model from GCS, caching after first load per executor."""
    if model_gcs_uri not in _model_cache:
        from transformers import pipeline as hf_pipeline
        from google.cloud import storage

        client = storage.Client()
        parts = model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]
        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)
        # Load tokenizer explicitly — DistilBERT doesn't accept token_type_ids
        # but some transformers versions (e.g. runtime 2.2) generate them
        from transformers import AutoTokenizer
        tokenizer = AutoTokenizer.from_pretrained(model_dir)
        if 'token_type_ids' in tokenizer.model_input_names:
            tokenizer.model_input_names.remove('token_type_ids')
        _model_cache[model_gcs_uri] = hf_pipeline(
            'sentiment-analysis', model=model_dir, tokenizer=tokenizer
        )
    return _model_cache[model_gcs_uri]


def run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--project', required=True)
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--model_gcs_uri', required=True)
    parser.add_argument('--model_name', default='distilbert')
    parser.add_argument('--temp_bucket', required=True)
    args = parser.parse_args()

    spark = SparkSession.builder.appName('batch-inference').getOrCreate()
    spark.conf.set('temporaryGcsBucket', args.temp_bucket)

    # Read from BigQuery
    df = spark.read.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.source_table}') \
        .load() \
        .select('id', 'text', 'category')

    # Define sentiment UDF with struct return type
    model_uri = args.model_gcs_uri

    schema = StructType([
        StructField('label', StringType()),
        StructField('score', FloatType()),
    ])

    @pandas_udf(schema)
    def sentiment_udf(texts: pd.Series) -> pd.DataFrame:
        pipe = get_model(model_uri)
        results = pipe(texts.tolist(), batch_size=32)
        return pd.DataFrame(results)

    # Apply inference and format output
    result_df = df.withColumn('prediction', sentiment_udf(col('text')))
    output_df = result_df.select(
        col('id'),
        col('text'),
        col('prediction.label').alias('label'),
        spark_round(col('prediction.score'), 6).alias('score'),
        lit(args.model_name).alias('model'),
    )

    # Write to BigQuery
    output_df.write.format('bigquery') \
        .option('table', f'{args.project}.{args.bq_dataset}.{args.output_table}') \
        .mode('overwrite') \
        .save()

    print(f'Results written to {args.project}.{args.bq_dataset}.{args.output_table}')


if __name__ == '__main__':
    run()
````

````python
%%writefile {DIR}/pipeline.py
"""Standalone Beam pipeline for sentiment batch inference."""
import argparse, json, logging, os, tempfile

logging.basicConfig(level=logging.INFO)

import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions, SetupOptions
from apache_beam.ml.inference.base import ModelHandler, PredictionResult, RunInference


class SentimentModelHandler(ModelHandler):
    def __init__(self, model_gcs_uri):
        self._model_gcs_uri = model_gcs_uri

    def load_model(self):
        from transformers import pipeline as hf_pipeline
        from google.cloud import storage
        client = storage.Client()
        parts = self._model_gcs_uri.replace('gs://', '').split('/', 1)
        bkt = client.bucket(parts[0])
        prefix = parts[1]
        model_dir = tempfile.mkdtemp()
        for blob in bkt.list_blobs(prefix=prefix):
            rel = blob.name[len(prefix):].lstrip('/')
            if rel:
                local = os.path.join(model_dir, rel)
                os.makedirs(os.path.dirname(local), exist_ok=True)
                blob.download_to_filename(local)
        return hf_pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)

    def run_inference(self, batch, model, inference_args=None):
        texts = [e['text'] for e in batch]
        results = model(texts)
        return [PredictionResult(example=e, inference=r) for e, r in zip(batch, results)]

    def batch_elements_kwargs(self):
        return {'min_batch_size': 1, 'max_batch_size': 32}


class PreprocessFn(beam.DoFn):
    def process(self, row):
        text = row.get('text', '')
        if isinstance(text, str):
            text = text.strip()
        if text:
            yield {'id': row.get('id', 0), 'text': text, 'category': row.get('category', '')}


class PostprocessFn(beam.DoFn):
    def process(self, prediction_result):
        element = prediction_result.example
        result = prediction_result.inference
        yield {
            'id': element['id'],
            'text': element['text'],
            'label': result['label'],
            'score': round(result['score'], 6),
            'model': 'distilbert',
        }


def run(argv=None):
    parser = argparse.ArgumentParser()
    # Note: --project is NOT declared here because DataflowPythonJobOp
    # auto-injects --project into the command. Declaring it here would
    # consume it via parse_known_args, preventing Beam from seeing it.
    parser.add_argument('--bq_dataset', required=True)
    parser.add_argument('--source_table', required=True)
    parser.add_argument('--output_table', required=True)
    parser.add_argument('--model_gcs_uri', required=True)
    known_args, pipeline_args = parser.parse_known_args(argv)

    # Get project from Beam pipeline options (auto-injected by the component)
    options = PipelineOptions(pipeline_args)
    options.view_as(SetupOptions).save_main_session = True
    project = options.get_all_options()['project']

    handler = SentimentModelHandler(known_args.model_gcs_uri)
    schema = 'id:INTEGER,text:STRING,label:STRING,score:FLOAT64,model:STRING'

    with beam.Pipeline(options=options) as p:
        (
            p
            | 'Read' >> beam.io.ReadFromBigQuery(
                query=f'SELECT id, text, category FROM `{project}.{known_args.bq_dataset}.{known_args.source_table}`',
                use_standard_sql=True,
            )
            | 'Preprocess' >> beam.ParDo(PreprocessFn())
            | 'Infer' >> RunInference(model_handler=handler)
            | 'Postprocess' >> beam.ParDo(PostprocessFn())
            | 'Write' >> beam.io.WriteToBigQuery(
                table=f'{project}:{known_args.bq_dataset}.{known_args.output_table}',
                schema=schema,
                write_disposition=beam.io.BigQueryDisposition.WRITE_TRUNCATE,
                create_disposition=beam.io.BigQueryDisposition.CREATE_IF_NEEDED,
            )
        )


if __name__ == '__main__':
    run()
````

````python
# Compile a KFP pipeline for the Airflow → KFP pattern
# This reuses the DataprocPySparkBatchOp pattern from Notebook 3
from kfp import dsl, compiler
from google_cloud_pipeline_components.v1.dataproc import DataprocPySparkBatchOp

@dsl.pipeline(
    name='airflow-dataproc-inference',
    description='Dataproc batch inference triggered by Airflow',
)
def airflow_inference_pipeline(
    project: str,
    region: str,
    script_gcs_path: str,
    model_gcs_uri: str,
    model_name: str,
    bq_dataset: str,
    source_table: str,
    output_table: str = 'kfp_predictions',
    temp_bucket: str = '',
):
    task = DataprocPySparkBatchOp(
        project=project,
        location=region,
        main_python_file_uri=script_gcs_path,
        runtime_config_version='2.2',
        runtime_config_properties={
            'spark.executor.instances': '2',
            'spark.driver.memory': '4g',
            'spark.executor.memory': '4g',
        },
        args=[
            '--project', project,
            '--bq_dataset', bq_dataset,
            '--source_table', source_table,
            '--output_table', output_table,
            '--model_gcs_uri', model_gcs_uri,
            '--model_name', model_name,
            '--temp_bucket', temp_bucket,
        ],
    )
    task.set_caching_options(False)

# Compile and upload
pipeline_json = f'{DIR}/airflow_inference_pipeline.json'
compiler.Compiler().compile(
    pipeline_func=airflow_inference_pipeline,
    package_path=pipeline_json,
)
print(f'Pipeline compiled to {pipeline_json}')

# Upload template to GCS
blob_path = f'{SERIES}/{EXPERIMENT}/airflow/airflow_inference_pipeline.json'
bucket.blob(blob_path).upload_from_filename(pipeline_json)
PIPELINE_TEMPLATE_GCS = f'gs://{GCS_BUCKET}/{blob_path}'
print(f'Pipeline template: {PIPELINE_TEMPLATE_GCS}')
````

````python
import time

def wait_for_dag_parsed(dag_id, timeout=300, poll_interval=15):
    """Wait until Airflow has parsed and registered the DAG."""
    print(f'Waiting for Airflow to parse DAG: {dag_id} ...')
    start = time.time()
    while time.time() - start < timeout:
        result = !gcloud composer environments run {COMPOSER_ENV} \
            --location {REGION} \
            dags list -- -o plain 2>/dev/null
        if dag_id in '\n'.join(result):
            print(f'DAG {dag_id} registered ({int(time.time() - start)}s)')
            return True
        time.sleep(poll_interval)
    raise TimeoutError(f'DAG {dag_id} not parsed after {timeout}s — check Airflow import errors')

def trigger_dag(dag_id, max_retries=3, retry_delay=20):
    """Trigger a DAG, retrying on scheduler-to-webserver sync delays."""
    for attempt in range(max_retries):
        result = !gcloud composer environments run {COMPOSER_ENV} \
            --location {REGION} \
            dags trigger -- {dag_id} 2>&1
        output = '\n'.join(result)
        if 'not found' not in output.lower():
            print(f'DAG {dag_id} triggered')
            return True
        if attempt < max_retries - 1:
            print(f'DAG {dag_id}: not ready (sync delay), retrying in {retry_delay}s ...')
            time.sleep(retry_delay)
    raise RuntimeError(f'Failed to trigger DAG {dag_id} after {max_retries} attempts')

def wait_for_dag(dag_id, timeout=1800, poll_interval=30):
    """Poll DAG run status until success or failure (checks latest run only)."""
    start = time.time()
    while time.time() - start < timeout:
        result = !gcloud composer environments run {COMPOSER_ENV} \
            --location {REGION} \
            dags list-runs -- -d {dag_id} -o table 2>/dev/null
        output = '\n'.join(result)
        if 'no data found' in output.lower():
            print(f'DAG {dag_id}: no run found — trigger may have failed')
            return 'not_triggered'
        # Check only the FIRST data row (latest run) — skip headers/separators
        for line in result:
            stripped = line.strip()
            if not stripped or stripped.startswith('=') or stripped.startswith('dag_id') or stripped.startswith('conf') or stripped.startswith('['):
                continue
            lower = stripped.lower()
            if '| success' in lower:
                print(f'DAG {dag_id}: SUCCESS ({int(time.time() - start)}s)')
                return 'success'
            elif '| failed' in lower:
                print(f'DAG {dag_id}: FAILED ({int(time.time() - start)}s)')
                return 'failed'
            elif '| running' in lower or '| queued' in lower:
                break  # still in progress, continue polling
            break  # unknown format, continue polling
        print(f'DAG {dag_id}: running ({int(time.time() - start)}s elapsed) ...')
        time.sleep(poll_interval)
    print(f'DAG {dag_id}: TIMEOUT after {timeout}s')
    return 'timeout'

print('Helper functions defined: wait_for_dag_parsed(), trigger_dag(), wait_for_dag()')
````

The simplest pattern. Airflow launches a Dataproc Serverless batch job directly using [`DataprocCreateBatchOperator`](https://airflow.apache.org/docs/apache-airflow-providers-google/stable/operators/cloud/dataproc.html). The operator submits the PySpark script and waits for completion.

> **Production tip:** Add `deferrable=True` to release the Airflow worker slot while waiting for long-running jobs. This requires a Medium+ Composer environment with a dedicated triggerer process.


````python
# Write DAG file with project-specific values injected
dag_dataproc = f'''"""DAG: Batch inference via Dataproc Serverless."""
from datetime import datetime
from airflow import models
from airflow.providers.google.cloud.operators.dataproc import DataprocCreateBatchOperator

PROJECT_ID = '{PROJECT_ID}'
REGION = '{REGION}'
SCRIPT_GCS = 'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/airflow/inference.py'
MODEL_GCS = '{MODELS["v1"]["artifact_uri"]}'
BQ_DATASET = '{BQ_DATASET}'
SOURCE_TABLE = '{BQ_TABLE}'
OUTPUT_TABLE = 'dataproc_predictions'

with models.DAG(
    'batch_inference_dataproc',
    schedule_interval=None,
    start_date=datetime(2026, 1, 1),
    catchup=False,
    tags=['batch-inference', 'dataproc'],
) as dag:

    run_inference = DataprocCreateBatchOperator(
        task_id='run_dataproc_inference',
        region=REGION,
        project_id=PROJECT_ID,
        batch={{
            'pyspark_batch': {{
                'main_python_file_uri': SCRIPT_GCS,
                'args': [
                    '--project', PROJECT_ID,
                    '--bq_dataset', BQ_DATASET,
                    '--source_table', SOURCE_TABLE,
                    '--output_table', OUTPUT_TABLE,
                    '--model_gcs_uri', MODEL_GCS,
                    '--model_name', 'distilbert',
                    '--temp_bucket', PROJECT_ID,
                ],
            }},
            'runtime_config': {{'version': '2.2'}},
        }},
        batch_id='airflow-dp-' + '{{{{ ts_nodash | lower }}}}',
    )
'''

with open(f'{DIR}/dag_dataproc.py', 'w') as f:
    f.write(dag_dataproc)
print(f'Written: {DIR}/dag_dataproc.py')
````

## DAG 2: Airflow → Dataflow

Uses [`BeamRunPythonPipelineOperator`](https://airflow.apache.org/docs/apache-airflow-providers-apache-beam/stable/operators.html) with `DataflowRunner`. This is the modern replacement for the deprecated `DataflowCreatePythonJobOperator` (removed in 2026 provider releases).

The operator uploads the Python pipeline script, creates a Dataflow job, and monitors it to completion.

> **Why a custom SDK container?** When a Beam pipeline has heavy dependencies (PyTorch \~2 GB), Dataflow downloads them via pip on the Airflow worker *before* submitting the job. On Composer SMALL environments, this blocks the worker long enough to trigger Airflow's zombie task detection — the scheduler kills the task thinking it's stuck. Building a [custom SDK container](https://cloud.google.com/dataflow/docs/guides/using-custom-containers) with dependencies pre-installed bypasses the local pip download entirely.


### Build Dataflow SDK Container

The SDK container is used by Dataflow *workers* — it must be based on the official `apache/beam_python3.x_sdk` image with the same Beam version installed in the Composer environment.


````python
# Get the Beam version installed in the Composer environment
result = !gcloud composer environments describe {COMPOSER_ENV} \
    --location {REGION} \
    --format="value(config.softwareConfig.imageVersion)"

IMAGE_VERSION = result[0].strip()
print(f'Composer image: {IMAGE_VERSION}')

# Get package list from the environment
pkg_result = !gcloud composer environments list-packages {COMPOSER_ENV} \
    --location {REGION} 2>/dev/null

BEAM_VERSION = None
for line in pkg_result:
    if line.strip().startswith('apache-beam') or line.strip().startswith('apache_beam'):
        parts = line.strip().split()
        if len(parts) >= 2:
            BEAM_VERSION = parts[1].strip()
            break

if not BEAM_VERSION:
    raise RuntimeError(f'Could not determine Beam version. Package list output: {pkg_result}')
print(f'Beam SDK version in Composer: {BEAM_VERSION}')
````

## DAG 3: Airflow → KFP Pipeline

The cleanest separation of concerns: Airflow handles *when* (scheduling, data dependencies), KFP handles *how* (ML pipeline logic, artifact tracking, lineage).

[`RunPipelineJobOperator`](https://airflow.apache.org/docs/apache-airflow-providers-google/stable/operators/cloud/vertex_ai.html) triggers a compiled KFP pipeline stored in GCS. The pipeline we compiled earlier uses `DataprocPySparkBatchOp` — so this is Airflow → KFP → Dataproc, a three-layer orchestration.

**When to use this pattern:**
- Team already uses KFP for ML pipeline management
- Need artifact lineage and experiment tracking (KFP provides this, Airflow doesn't)
- ML pipeline has multiple steps (preprocess → inference → postprocess) that benefit from KFP's DAG visualization


````python
dag_kfp = f'''"""DAG: Batch inference via Vertex AI Pipelines (KFP)."""
from datetime import datetime
from airflow import models
from airflow.providers.google.cloud.operators.vertex_ai.pipeline_job import RunPipelineJobOperator

PROJECT_ID = '{PROJECT_ID}'
REGION = '{REGION}'
PIPELINE_TEMPLATE = '{PIPELINE_TEMPLATE_GCS}'
SCRIPT_GCS = 'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/airflow/inference.py'
MODEL_GCS = '{MODELS["v1"]["artifact_uri"]}'
BQ_DATASET = '{BQ_DATASET}'
SOURCE_TABLE = '{BQ_TABLE}'

with models.DAG(
    'batch_inference_kfp',
    schedule_interval=None,
    start_date=datetime(2026, 1, 1),
    catchup=False,
    tags=['batch-inference', 'kfp'],
) as dag:

    run_pipeline = RunPipelineJobOperator(
        task_id='run_kfp_pipeline',
        project_id=PROJECT_ID,
        region=REGION,
        display_name='airflow-kfp-inference',
        template_path=PIPELINE_TEMPLATE,
        parameter_values={{
            'project': PROJECT_ID,
            'region': REGION,
            'script_gcs_path': SCRIPT_GCS,
            'model_gcs_uri': MODEL_GCS,
            'model_name': 'distilbert',
            'bq_dataset': BQ_DATASET,
            'source_table': SOURCE_TABLE,
            'output_table': 'kfp_predictions',
            'temp_bucket': PROJECT_ID,
        }},
    )
'''

with open(f'{DIR}/dag_kfp.py', 'w') as f:
    f.write(dag_kfp)
print(f'Written: {DIR}/dag_kfp.py')
````

### Airflow → KFP vs Direct

| Aspect | Airflow → KFP | Airflow Direct |
|--------|:-:|:-:|
| ML artifact lineage | KFP tracks artifacts, models, metrics | No tracking |
| Pipeline visualization | KFP DAG + Airflow DAG | Airflow DAG only |
| Multi-step ML logic | KFP handles preprocess → infer → postprocess | Gets complex in a single DAG |
| Overhead | Two orchestrators | One orchestrator |
| Best for | ML-first teams with existing KFP pipelines | Simple batch jobs, Airflow-native teams |

## DAG 4: Airflow → Vertex AI Batch Prediction

This pattern uses [`CreateBatchPredictionJobOperator`](https://airflow.apache.org/docs/apache-airflow-providers-google/stable/operators/cloud/vertex_ai.html) to launch a Vertex AI Batch Prediction job directly from Airflow.

Unlike patterns 1-3 which use standalone inference scripts, Vertex AI Batch Prediction requires:
1. A **serving container** in Artifact Registry that hosts the model as an API
2. A **model registered** in the Vertex AI Model Registry pointing to the container and model artifacts

We build both below, then deploy the DAG.


### Build Prediction Container

The prediction container is a FastAPI app that Vertex AI scales horizontally for batch inference. Each replica downloads model files from GCS (via `AIP_STORAGE_URI`), loads the HuggingFace model, and serves predictions on the `/predict` endpoint.

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model from GCS at startup (AIP_STORAGE_URI set by Vertex AI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)
for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']

    texts = []
    for inst in instances:
        if isinstance(inst, str):
            texts.append(inst)
        elif isinstance(inst, dict):
            texts.append(inst.get('text', str(inst)))
        else:
            texts.append(str(inst))

    results = classifier(texts)
    predictions = [
        {'label': r['label'], 'score': round(r['score'], 6)}
        for r in results
    ]
    return {'predictions': predictions}
````

## Production Patterns

The DAGs above demonstrate the mechanics. In production, Airflow adds scheduling, retry, backfill, and multi-model coordination.


### Scheduling & Catchup

```python
with models.DAG(
    'daily_inference',
    schedule_interval='0 2 * * *',    # Run daily at 2 AM
    start_date=datetime(2026, 4, 1),
    catchup=False,                     # Don't backfill past dates on first deploy
    max_active_runs=1,                 # One run at a time
) as dag:
    ...
```

- `schedule_interval` accepts cron expressions or Airflow presets (`@daily`, `@hourly`)
- `catchup=False` prevents Airflow from creating runs for every missed interval since `start_date`
- `max_active_runs=1` prevents overlapping inference jobs that could compete for compute resources


### Backfill — Airflow's Killer Feature for Batch ML

Missed a day? A model was wrong and you need to re-score last week?

```bash
# Re-run inference for April 15-17
gcloud composer environments run my-env --location us-central1 \
    dags backfill -- -s 2026-04-15 -e 2026-04-17 daily_inference
```

Airflow creates one DAG run per date, each with its own execution date. The DAG code uses `{{ ds }}` (execution date template) to:
- Read the correct input partition: `WHERE date = '{{ ds }}'`
- Write to the correct output partition: `output_{{ ds_nodash }}`
- Ensure idempotency: re-running the same date overwrites previous results

KFP has no native equivalent. Without Airflow, backfill means writing custom scripts to loop over dates and submit jobs manually.


### Multi-Model with TaskGroups

Run both models in parallel, then compare results:

```python
from airflow.utils.task_group import TaskGroup

with TaskGroup('model_inference') as inference:
    distilbert = DataprocCreateBatchOperator(
        task_id='distilbert_inference',
        batch={...},  # model_gcs_uri → v1
    )
    bert = DataprocCreateBatchOperator(
        task_id='bert_inference',
        batch={...},  # model_gcs_uri → v2
    )

compare = BigQueryInsertJobOperator(
    task_id='compare_models',
    configuration={'query': {'query': '''
        SELECT a.id, a.text, a.label AS distilbert, b.label AS bert
        FROM predictions_v1 a JOIN predictions_v2 b USING (id)
    '''}},
)

inference >> compare
```

TaskGroups are visual only — they group tasks in the Airflow UI but don't change execution behavior. Both models run in parallel because there's no dependency between them.


### Retry & Alerting

```python
default_args = {
    'retries': 2,
    'retry_delay': timedelta(minutes=5),
    'on_failure_callback': alert_on_failure,  # Slack, email, PagerDuty
    'sla': timedelta(hours=2),                # Alert if job exceeds 2 hours
}
```

- **Retries**: Batch jobs fail sometimes — quota limits, transient network errors, preemptible worker loss. 2 retries with 5-minute delays handles most transient failures.
- **SLA**: `sla` flags tasks that exceed a time limit. Critical for ML pipelines with downstream consumers waiting for fresh predictions.
- **Callbacks**: `on_failure_callback` fires when a task fails all retries. Send to Slack, PagerDuty, or Cloud Monitoring.


### Idempotent DAGs

DAGs should be idempotent: running the same date twice produces the same result without duplicating data.

**Pattern:** Write to date-partitioned BigQuery tables, overwrite partition on re-run:

```python
output_table = f'predictions_{ds_nodash}'  # predictions_20260418

# In the inference script:
df.write.format('bigquery') \
    .option('table', f'{project}.{dataset}.{output_table}') \
    .mode('overwrite') \
    .save()
```

**Anti-pattern:** Append-only writes that duplicate data on retry. If a job fails halfway and retries, the append includes both the partial and complete results.


### Decision Framework

| Need | Airflow → KFP | Airflow Direct (Dataproc/Dataflow) |
|------|:-:|:-:|
| ML artifact lineage | Yes (KFP tracks) | No |
| Multi-step ML pipeline | KFP handles it | Gets complex in DAG |
| Team already has KFP | Leverage existing pipelines | Unnecessary layer |
| Simple batch predict | Extra overhead | Clean and simple |
| Backfill | Both | Both |
| Cross-system deps | Both (Airflow handles) | Both |
| Monitoring | Airflow UI + KFP UI | Airflow UI only |
