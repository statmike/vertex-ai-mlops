# Vertex AI Private Endpoint With Private Service Connect (PSC)

Deploy HuggingFace sentiment analysis models to a [Private Endpoint using Private Service Connect (PSC)](https://cloud.google.com/vertex-ai/docs/predictions/private-service-connect) on Vertex AI. This notebook demonstrates:

- Building a custom FastAPI prediction container with Cloud Build
- Creating a PSC-enabled private endpoint
- Setting up private networking (reserved IP, forwarding rule)
- Using a GCE VM as a test client within the VPC to reach the private IP
- Deploying, swapping, and undeploying models while proving the private connection remains stable

**The key insight:** With PSC, the endpoint gets a stable private IP address in your VPC. Even as you deploy new models, shift traffic, and undeploy old models, the private IP and connection remain uninterrupted — the underlying service never goes offline.

**Story arc:**
1. Deploy DistilBERT (100% traffic) — serve predictions via private IP
2. Deploy BERT alongside (0% traffic) — predictions still served by DistilBERT
3. Shift 100% traffic to BERT — same private IP, now BERT results
4. Undeploy DistilBERT — same private IP still works (endpoint and PSC connection are stable)

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params)
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params)

## Download HuggingFace Models

Download two sentiment analysis models from HuggingFace and upload them to GCS. Both models are fine-tuned on the [SST-2](https://huggingface.co/datasets/stanfordnlp/sst2) dataset for binary sentiment classification (POSITIVE/NEGATIVE).

- **v1 (DistilBERT):** Smaller, faster — good for initial deployment
- **v2 (BERT):** Larger, more accurate — the model we'll swap to

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'
    
    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue
    
    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)
        
        # upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    # store the GCS URI
    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

````python
for version, info in MODELS.items():
    blobs = list(bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/models/{version}'))
    print(f'{version} ({info["display_name"]}): {len(blobs)} files at {info["artifact_uri"]}')
````

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)
for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']
    results = classifier(instances)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
````

## Create PSC Endpoint and Deploy First Model

Create a [private endpoint with Private Service Connect (PSC)](https://cloud.google.com/vertex-ai/docs/predictions/private-service-connect), then deploy the first model.

**Why deploy before setting up networking?** A PSC endpoint is a logical resource — no serving infrastructure is provisioned until a model is actually deployed. The [service attachment](https://cloud.google.com/vpc/docs/private-service-connect#service-attachments) that the forwarding rule needs to target only exists once there are running resources behind it. So the flow is:

1. Create the PSC endpoint (logical resource, no compute yet)
2. Deploy a model (Vertex AI provisions serving nodes and creates the service attachment)
3. Retrieve the service attachment and set up the forwarding rule

````python
endpoint = aiplatform.PrivateEndpoint.create(
    display_name=f'{EXPERIMENT}-psc',
    private_service_connect_config=aiplatform.PrivateEndpoint.PrivateServiceConnectConfig(
        project_allowlist=[PROJECT_ID],
    ),
    labels={'series': SERIES, 'experiment': EXPERIMENT},
)
print(f'Endpoint: {endpoint.resource_name}')
````

### Helper: Send Predictions via VM

Define a helper function that runs `curl` on the GCE VM to send prediction requests to the private PSC IP address.

````python
def predict_via_vm(instances, endpoint_id=None, psc_ip=None):
    """Send a prediction request to the PSC private endpoint via the GCE VM."""
    if psc_ip is None:
        psc_ip = PSC_IP
    if endpoint_id is None:
        endpoint_id = endpoint.name.split('/')[-1]
    
    payload = json.dumps({'instances': instances})
    
    curl_cmd = (
        f'curl -s -X POST '
        f'-H "Content-Type: application/json" '
        f'-H "Authorization: Bearer $(gcloud auth print-access-token)" '
        f'-d \'{payload}\' '
        f'https://{psc_ip}/v1/projects/{PROJECT_ID}/locations/{REGION}/endpoints/{endpoint_id}:predict '
        f'-k'
    )
    
    result = subprocess.run(
        [
            'gcloud', 'compute', 'ssh', VM_NAME,
            f'--zone={REGION}-a',
            f'--project={PROJECT_ID}',
            '--command', curl_cmd,
            '--quiet',
        ],
        capture_output=True, text=True,
    )
    
    # Try to parse stdout regardless of return code
    # (SSH key warnings go to stderr but don't affect the response)
    if result.stdout.strip():
        try:
            return json.loads(result.stdout)
        except json.JSONDecodeError:
            print(f'Unexpected response: {result.stdout.strip()}')
            return None
    
    # No stdout — show diagnostics
    print(f'No response from endpoint (exit code {result.returncode})')
    if result.stderr.strip():
        lines = [l for l in result.stderr.strip().split('\n')
                 if 'WARNING:' not in l and 'known hosts' not in l]
        if lines:
            print('\n'.join(lines))
    return None

# Test sentences for sentiment analysis
TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]
````

## Test: Predictions via PSC Private IP (DistilBERT)

Send prediction requests from the GCE VM through the private IP. All responses should come from DistilBERT.

````python
print(f'Sending predictions to private IP: {PSC_IP}')
print(f'Expected model: DistilBERT (v1)\n')

response = predict_via_vm(TEST_INSTANCES)
if response:
    for instance, pred in zip(TEST_INSTANCES, response['predictions']):
        print(f'  "{instance[:60]}..."')
        print(f'    => {pred["label"]} (score: {pred["score"]})')
        print()
````

## Deploy BERT (v2) Alongside — 0% Traffic

Deploy the larger BERT model alongside DistilBERT, initially receiving no traffic. This lets the model warm up before we shift traffic to it.

````python
endpoint.deploy(
    model=vertex_model_v2,
    deployed_model_display_name='bert-sst2',
    machine_type='n1-standard-4',
    min_replica_count=1,
    max_replica_count=2,
    traffic_percentage=0,
)
print(f'Deployed BERT alongside DistilBERT with 0% traffic')
print(f'Traffic split: {endpoint.traffic_split}')
````

### Test: Predictions Still Served by DistilBERT

Even though BERT is now deployed, all traffic still goes to DistilBERT. The private IP hasn't changed.

````python
print(f'Sending predictions to private IP: {PSC_IP}')
print(f'Expected model: DistilBERT (v1) — BERT is deployed but at 0% traffic\n')

response = predict_via_vm(TEST_INSTANCES)
if response:
    for instance, pred in zip(TEST_INSTANCES, response['predictions']):
        print(f'  "{instance[:60]}..."')
        print(f'    => {pred["label"]} (score: {pred["score"]})')
        print()
````

## Shift 100% Traffic to BERT (v2)

Now shift all traffic from DistilBERT to BERT. The private IP remains the same — only the model serving predictions changes.

````python
print(f'Current traffic split: {endpoint.traffic_split}')

# Build new traffic split: 100% to BERT (v2), 0% to DistilBERT (v1)
new_traffic_split = {}
for deployed_model in endpoint.list_models():
    if deployed_model.display_name == 'bert-sst2':
        new_traffic_split[deployed_model.id] = 100
    else:
        new_traffic_split[deployed_model.id] = 0

print(f'New traffic split: {new_traffic_split}')
endpoint.update(traffic_split=new_traffic_split)
print(f'Updated traffic split: {endpoint.traffic_split}')
````

### Test: Predictions Now Served by BERT

Same private IP, but now the responses come from BERT. Notice the scores differ from DistilBERT — it's a different (larger) model.

> **Label difference:** BERT (v2) returns `LABEL_0`/`LABEL_1` instead of `POSITIVE`/`NEGATIVE` because the [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) model config uses numeric label IDs while DistilBERT's config has named labels. The different labels actually help confirm the traffic shift worked. To control label names, set `id2label` in the model's `config.json` (e.g., `{"0": "NEGATIVE", "1": "POSITIVE"}`) before uploading to GCS.

````python
print(f'Sending predictions to private IP: {PSC_IP}')
print(f'Expected model: BERT (v2) — traffic shifted from DistilBERT\n')

response = predict_via_vm(TEST_INSTANCES)
if response:
    for instance, pred in zip(TEST_INSTANCES, response['predictions']):
        print(f'  "{instance[:60]}..."')
        print(f'    => {pred["label"]} (score: {pred["score"]})')
        print()
````

## Undeploy DistilBERT — Private IP Still Stable

Remove the DistilBERT model entirely from the endpoint. The endpoint continues to serve via the same private IP — no disruption to the PSC connection.

````python
# Find and undeploy the DistilBERT model (which now has 0% traffic)
for deployed_model in endpoint.list_models():
    if deployed_model.display_name == 'distilbert-sst2':
        print(f'Undeploying {deployed_model.display_name} (id: {deployed_model.id})')
        endpoint.undeploy(deployed_model_id=deployed_model.id)
        break

print(f'\nRemaining deployed models:')
for m in endpoint.list_models():
    print(f'  {m.display_name} — traffic: {endpoint.traffic_split.get(m.id, 0)}%')
````

### Test: Private IP Still Works After Undeploy

This is the key proof: the private IP address and PSC connection remain stable even after undeploying a model. The endpoint never went offline.

````python
print(f'Sending predictions to private IP: {PSC_IP}')
print(f'Expected model: BERT (v2) — DistilBERT has been undeployed')
print(f'Key point: same private IP, connection never interrupted\n')

response = predict_via_vm(TEST_INSTANCES)
if response:
    for instance, pred in zip(TEST_INSTANCES, response['predictions']):
        print(f'  "{instance[:60]}..."')
        print(f'    => {pred["label"]} (score: {pred["score"]})')
        print()
    print('Private endpoint remained stable through deploy, traffic shift, and undeploy!')
````
