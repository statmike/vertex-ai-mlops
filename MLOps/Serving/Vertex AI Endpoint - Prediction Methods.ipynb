# Vertex AI Endpoint - Prediction Methods

A comprehensive reference for every way to send prediction requests to a [Vertex AI Endpoint](https://cloud.google.com/vertex-ai/docs/predictions/get-online-predictions) — Python SDK (high-level and gapic), REST API in multiple languages (Python, Node.js, Java, Go), gRPC, streaming (SSE), `gcloud` CLI, and async concurrency patterns.

**Story arc:**
1. Deploy a single HuggingFace sentiment model to a dedicated public endpoint
2. Demonstrate every prediction method against the same endpoint
3. Benchmark sync vs async concurrency patterns
4. Explain how methods differ for other endpoint types (shared public, PSC private)

**Why a dedicated public endpoint?** It supports ALL prediction methods — gRPC, streaming (SSE), dedicated DNS URL, `directPredict`, and standard REST. The compatibility table below shows which methods work with each endpoint type.

**Prior art:** This notebook modernizes and consolidates [05Tools - Prediction - Online.ipynb](../../05%20-%20TensorFlow/05Tools%20-%20Prediction%20-%20Online.ipynb), which covers the Python SDK layers in detail. We simplify async code (the SDK now has built-in `predict_async()`), consolidate redundant client sections, and add gRPC, streaming, multi-language REST, and endpoint type compatibility.

### Endpoint Type Compatibility

Not all prediction methods are available on all endpoint types. This notebook uses a **dedicated public endpoint** which supports everything. Each section below notes its endpoint compatibility.

| Method | Dedicated Public | Shared Public | PSC Private | VPC Peering |
|--------|:---:|:---:|:---:|:---:|
| `predict` / `rawPredict` (SDK, REST, gRPC) | Yes | Yes | Yes | Yes |
| `streamRawPredict` / SSE | **Yes** | No | **Yes** | No |
| Dedicated endpoint direct URL | **Yes** | No | No | No |
| `directPredict` / `directRawPredict` | **Yes** | No | **Yes** | No |
| gRPC transport | **Yes** | No | **Yes** | No |
| `gcloud ai endpoints predict` | Yes | Yes | Yes | Yes |
| Batch Prediction | Yes | Yes | No | No |

> See [readme.md](./readme.md) for detailed endpoint type descriptions and the full comparison table. For PSC endpoint setup, see [Vertex AI Private Endpoint With PSC](./Vertex%20AI%20Private%20Endpoint%20With%20PSC.ipynb).

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)
for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']
    results = classifier(instances)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
````

## Python SDK (High-Level) — `google.cloud.aiplatform`

The [`aiplatform.Endpoint`](https://cloud.google.com/python/docs/reference/aiplatform/latest/google.cloud.aiplatform.Endpoint) object provides the simplest prediction API. This is what most Python users should reach for first.

The high-level SDK methods `predict()`, `raw_predict()`, and `stream_raw_predict()` **automatically handle dedicated endpoint routing** — they detect `dedicated_endpoint_enabled=True` and route through the dedicated DNS. No special flags needed.

**Endpoint compatibility:** `predict` and `raw_predict` work on all endpoint types. `stream_raw_predict` requires dedicated or PSC.

### `endpoint.predict()` — Standard Prediction

Accepts a list of instances (Python objects), returns a [`Prediction`](https://cloud.google.com/python/docs/reference/aiplatform/latest/google.cloud.aiplatform.models.Prediction) namedtuple with `.predictions`, `.deployed_model_id`, `.model_version_id`, and `.model_resource_name`.

````python
response = endpoint.predict(instances=TEST_INSTANCES)

print(f'Type: {type(response).__name__}')
print(f'Deployed model ID: {response.deployed_model_id}')
print(f'Model version: {response.model_version_id}')
print(f'Model resource: {response.model_resource_name}')
print()
for instance, pred in zip(TEST_INSTANCES, response.predictions):
    print(f'  "{instance[:50]}..." => {pred["label"]} ({pred["score"]})')
````

### `endpoint.raw_predict()` — Raw Prediction

Send an arbitrary HTTP body and headers. Returns a [`requests.Response`](https://docs.python-requests.org/en/latest/api/#requests.Response) object. Useful when you need custom headers, non-standard payloads, or want the raw HTTP response.

````python
body = json.dumps({'instances': TEST_INSTANCES}).encode('utf-8')
headers = {'Content-Type': 'application/json'}

response = endpoint.raw_predict(body=body, headers=headers)

print(f'Status: {response.status_code}')
predictions = json.loads(response.text)
for instance, pred in zip(TEST_INSTANCES, predictions['predictions']):
    print(f'  "{instance[:50]}..." => {pred["label"]} ({pred["score"]})')
````

### `endpoint.predict_async()` — Async Prediction

Built-in async coroutine — returns the same `Prediction` namedtuple. Use with `await` in Jupyter (which runs an asyncio event loop). No manual asyncio wrapping needed.

> **SDK limitation (as of v1.143.0):** `predict_async()` routes through the regional gRPC endpoint (`aiplatform.googleapis.com`), which **rejects dedicated endpoints** with `FailedPrecondition`. The sync `predict()` has special code to route through the dedicated DNS, but `predict_async()` does not.
>
> **Workaround:** Use `asyncio.to_thread(endpoint.predict, ...)` to run the sync method — which handles dedicated routing correctly — in a thread-pool executor. This provides real concurrency since the GIL is released during I/O. This works on shared/PSC/VPC endpoints too.
>
> **Deep fix:** See the [Appendix](#appendix-making-gapic-work-with-dedicated-endpoints) for a monkey-patch that makes `predict_async()` work directly on dedicated endpoints.

> **Note:** There is no `raw_predict_async()` on the high-level `Endpoint`. For async raw predictions, use the gapic `PredictionServiceAsyncClient` shown in the next section, or `httpx.AsyncClient` with the dedicated DNS URL.

````python
# predict_async() fails on dedicated endpoints — routes through regional API, not dedicated DNS
# Workaround: wrap the sync predict() in asyncio.to_thread
response = await asyncio.to_thread(endpoint.predict, instances=TEST_INSTANCES)

print(f'Type: {type(response).__name__}')
for instance, pred in zip(TEST_INSTANCES, response.predictions):
    print(f'  "{instance[:50]}..." => {pred["label"]} ({pred["score"]})')
````

### `endpoint.stream_raw_predict()` — Streaming (SSE)

Returns a sync iterator of response chunks. For models that produce output incrementally (e.g., LLMs generating tokens), each chunk arrives as it's ready. Our simple sentiment model returns a single chunk, but the method mechanics are the same.

**Endpoint compatibility:** Dedicated public and PSC private only.

## Dedicated Endpoint Routing

Dedicated endpoints get their own DNS entry that **bypasses the regional API router** and routes directly to the endpoint's serving infrastructure. The regional API (`aiplatform.googleapis.com`) **rejects prediction requests** for dedicated endpoints — all requests must go through the dedicated DNS.

**How the SDK handles this:**
- `predict()`, `raw_predict()`, `stream_raw_predict()` — **automatic.** The SDK detects `dedicated_endpoint_enabled=True` and sends plain HTTP POST to the dedicated DNS via `AuthorizedSession`, bypassing the gapic layer entirely.
- `predict_async()` — **broken (SDK bug as of v1.143.0).** Always routes through the regional gRPC endpoint. Use `asyncio.to_thread(endpoint.predict, ...)` as a workaround.

**What doesn't work with dedicated endpoints:**
- **gapic clients** (`aiplatform_v1.PredictionServiceClient`) — neither transport works with the dedicated DNS. gRPC returns `501 MethodNotImplemented`. REST appends a `$alt` query parameter that the dedicated DNS rejects with `400 Bad Request`. The gapic REST transport always adds this parameter — it cannot be disabled.

**What works with a workaround:**
- **gcloud CLI** — by default routes through the regional API (rejected). Fix: `gcloud config set api_endpoint_overrides/aiplatform https://{dedicated_dns}/` to route gcloud through the dedicated DNS. Reset after prediction calls.

**Direct HTTP** (requests, httpx, curl) to the dedicated DNS URL works with no special configuration:

```
Regional (shared/PSC/VPC):  https://{REGION}-aiplatform.googleapis.com/v1/{resource_name}:predict
Dedicated:                  https://{dedicated_endpoint_dns}/v1/{resource_name}:predict
```

All REST and curl examples in this notebook use the dedicated DNS. Comments show what to change for other endpoint types.

## Sync vs Async Concurrency

When you need to make many prediction requests, the biggest speedup comes from **async concurrency** — sending multiple requests simultaneously rather than waiting for each one to complete.

**Why not `asyncio.to_thread(endpoint.predict)`?** While it works for individual calls (shown above), it doesn't achieve real parallelism for benchmarking. The SDK's `endpoint.predict()` shares an internal `AuthorizedSession` and holds the GIL during JSON serialization and response parsing, which serializes concurrent threads. For true async concurrency, use `httpx.AsyncClient` with the dedicated DNS URL — each request is genuinely non-blocking.

This section adapts the concurrency patterns from [05Tools - Prediction - Online.ipynb](../../05%20-%20TensorFlow/05Tools%20-%20Prediction%20-%20Online.ipynb) using `httpx` for async and the high-level SDK for sync.

````python
# Create a larger test set for benchmarking
BENCH_INSTANCES = TEST_INSTANCES * 34  # 102 instances
print(f'Benchmark instances: {len(BENCH_INSTANCES)}')
````

### Synchronous Batching

````python
def sync_predictions(instances, batch_size=1):
    predictions = []
    n_requests = (len(instances) + batch_size - 1) // batch_size
    start = time.perf_counter()
    for i in range(0, len(instances), batch_size):
        response = endpoint.predict(instances=instances[i:i+batch_size])
        predictions.extend(response.predictions)
    elapsed = time.perf_counter() - start
    return predictions, elapsed, n_requests
````

````python
print('Synchronous batching (sequential requests via SDK):')
print(f'  {"batch_size":>10s} | {"requests":>8s} | {"time":>7s}')
print(f'  {"-"*10} | {"-"*8} | {"-"*7}')
for bs in [1, 3, 10, 50]:
    _, elapsed, n_req = sync_predictions(BENCH_INSTANCES, batch_size=bs)
    print(f'  {bs:>10d} | {n_req:>8d} | {elapsed:>6.2f}s')

baseline_time = sync_predictions(BENCH_INSTANCES, batch_size=1)[1]
````

### Async Concurrency

Use `httpx.AsyncClient` with `asyncio.gather()` to send multiple requests concurrently. An `asyncio.Semaphore` limits the number of in-flight requests to avoid overwhelming the endpoint. Unlike `asyncio.to_thread`, each `httpx` request is genuinely non-blocking — no GIL contention.

````python
async def async_predictions(instances, batch_size=1, max_concurrent=10):
    credentials.refresh(auth_req)
    semaphore = asyncio.Semaphore(max_concurrent)
    predictions = [None] * len(instances)
    n_requests = (len(instances) + batch_size - 1) // batch_size

    async def predict_batch(client, start_idx):
        async with semaphore:
            batch = instances[start_idx:start_idx + batch_size]
            response = await client.post(
                PREDICT_URL,
                json={'instances': batch},
                headers={
                    'Authorization': f'Bearer {credentials.token}',
                    'Content-Type': 'application/json',
                },
            )
            preds = response.json()['predictions']
            predictions[start_idx:start_idx + batch_size] = preds

    async with httpx.AsyncClient() as client:
        tasks = [
            asyncio.create_task(predict_batch(client, i))
            for i in range(0, len(instances), batch_size)
        ]
        start = time.perf_counter()
        await asyncio.gather(*tasks)
        elapsed = time.perf_counter() - start

    return predictions, elapsed, n_requests
````

````python
print(f'Async concurrency (parallel requests via httpx):')
print(f'  {"batch_size":>10s} | {"concurrency":>11s} | {"requests":>8s} | {"time":>7s} | {"vs sync":>7s}')
print(f'  {"-"*10} | {"-"*11} | {"-"*8} | {"-"*7} | {"-"*7}')
for bs, concur in [(1, 1), (1, 10), (1, 20), (3, 10), (10, 10)]:
    _, elapsed, n_req = await async_predictions(BENCH_INSTANCES, batch_size=bs, max_concurrent=concur)
    speedup = baseline_time / elapsed
    print(f'  {bs:>10d} | {concur:>11d} | {n_req:>8d} | {elapsed:>6.2f}s | {speedup:>5.1f}x')
````

### Takeaways

- **Batch size** reduces the number of API round trips — bigger batches = fewer requests = faster (even with sync)
- **Async concurrency** sends multiple requests in parallel — the biggest speedup for many small requests
- **Combining both** (batch size + concurrency) gives the best throughput
- The `asyncio.Semaphore` prevents overwhelming the endpoint with too many simultaneous requests
- Use `httpx.AsyncClient` for genuine async parallelism — `asyncio.to_thread(endpoint.predict)` serializes through the GIL and doesn't achieve real concurrency
- The **"vs sync" column** compares each configuration against sync batch_size=1 (the baseline)
- For production, add [retry logic](#error-handling-and-retries) to handle transient 503/429 errors

## Error Handling and Retries

Common errors when sending many requests:
- **`503 Model server connection error`** — too many concurrent requests overwhelmed the endpoint
- **`429 Quota exceeded`** — regional QPM limit reached ([quotas docs](https://cloud.google.com/vertex-ai/docs/quotas))

Pattern: exponential backoff with a retry limit.

````python
async def predict_with_retry(client, instances, max_retries=5):
    credentials.refresh(auth_req)
    for attempt in range(max_retries + 1):
        try:
            response = await client.post(
                PREDICT_URL,
                json={'instances': instances},
                headers={
                    'Authorization': f'Bearer {credentials.token}',
                    'Content-Type': 'application/json',
                },
            )
            response.raise_for_status()
            return response.json()
        except Exception as e:
            if attempt == max_retries:
                raise
            wait = min(2 ** attempt, 32)
            print(f'  Attempt {attempt + 1} failed ({type(e).__name__}), retrying in {wait}s...')
            await asyncio.sleep(wait)

# Test with normal input — should succeed on first attempt
async with httpx.AsyncClient() as client:
    result = await predict_with_retry(client, TEST_INSTANCES)
    print(f'Success: {result["predictions"][0]["label"]}')
````

````python
async def robust_async_predictions(instances, batch_size=1, max_concurrent=10, max_retries=5):
    credentials.refresh(auth_req)
    semaphore = asyncio.Semaphore(max_concurrent)
    predictions = [None] * len(instances)

    async def predict_batch(client, start_idx):
        async with semaphore:
            batch = instances[start_idx:start_idx + batch_size]
            for attempt in range(max_retries + 1):
                try:
                    response = await client.post(
                        PREDICT_URL,
                        json={'instances': batch},
                        headers={
                            'Authorization': f'Bearer {credentials.token}',
                            'Content-Type': 'application/json',
                        },
                    )
                    response.raise_for_status()
                    preds = response.json()['predictions']
                    predictions[start_idx:start_idx + batch_size] = preds
                    return
                except Exception as e:
                    if attempt == max_retries:
                        raise
                    await asyncio.sleep(min(2 ** attempt, 32))

    async with httpx.AsyncClient() as client:
        tasks = [asyncio.create_task(predict_batch(client, i))
                 for i in range(0, len(instances), batch_size)]
        await asyncio.gather(*tasks)
    return predictions

# Run with retry support
results = await robust_async_predictions(BENCH_INSTANCES, batch_size=3, max_concurrent=10)
print(f'Completed {len(results)} predictions with retry support')
````

## Appendix: Making gapic Work with Dedicated Endpoints

> **These are unsupported workarounds for educational purposes only.** They rely on monkey-patching internal SDK behavior that can change without notice in any SDK update. For production code, use the high-level SDK (`endpoint.predict()`) or direct HTTP (`requests`/`httpx`/`curl`) with the dedicated DNS URL.

The gapic `PredictionServiceClient` fails on dedicated endpoints for three independent reasons:

1. **`$alt` query parameter** — the gapic REST transport always appends `$alt=json;enum-encoding=int` to every request URL. The dedicated DNS doesn't understand this parameter and returns **`400 Bad Request`**.

2. **HttpBody request wrapping** — for `raw_predict` and `stream_raw_predict`, the transport serializes the `RawPredictRequest` protobuf using `MessageToJson`, producing `{"httpBody": {"contentType": "...", "data": "<base64>"}}`. The dedicated DNS expects the raw body, so it returns **`500 Internal Server Error`**.

3. **HttpBody response parsing** — the transport parses the raw JSON response (`{"predictions": [...]}`) with `json_format.Parse` into an `HttpBody` protobuf. Since `predictions` isn't an HttpBody field, it gets silently ignored (`ignore_unknown_fields=True`) and `response.data` comes back **empty**.

We can fix all three with targeted monkey-patches. We can also fix `endpoint.predict_async()` by swapping its internal gapic client to route through the dedicated DNS.

**What remains unfixable:** The gRPC transport returns **`501 MethodNotImplemented`** — the dedicated DNS routes to an HTTP/REST serving frontend, not a gRPC server. This is an infrastructure limitation, not a client-side issue.

### Patch 1: Strip `$alt` Query Parameter

Every inner class in the gapic REST transport base has a `_get_query_params_json` method that hardcodes `$alt=json;enum-encoding=int`. We patch all of them at once to remove this parameter. This fixes `predict()`, `raw_predict()`, and all other REST methods for dedicated endpoints.

````python
from google.cloud.aiplatform_v1.services.prediction_service.transports.rest_base import (
    _BasePredictionServiceRestTransport,
)

_patched_originals = {}  # save originals for cleanup

for cls_name in dir(_BasePredictionServiceRestTransport):
    if cls_name.startswith('_Base') and cls_name != '_BasePredictionServiceRestTransport':
        inner_cls = getattr(_BasePredictionServiceRestTransport, cls_name)
        if hasattr(inner_cls, '_get_query_params_json'):
            original = inner_cls._get_query_params_json
            _patched_originals[f'{cls_name}._get_query_params_json'] = (inner_cls, original)

            def make_patched(orig):
                @staticmethod
                def patched(transcoded_request):
                    result = orig(transcoded_request)
                    result.pop('$alt', None)
                    return result
                return patched

            inner_cls._get_query_params_json = make_patched(original)

print(f'Patched {len(_patched_originals)} inner classes to strip $alt parameter')
````

````python
# With the $alt patch, gapic REST predict() now works with dedicated endpoints
predictor = aiplatform_v1.PredictionServiceClient(
    transport='rest',
    client_options={'api_endpoint': endpoint.dedicated_endpoint_dns},
)

response = predictor.predict(
    endpoint=endpoint.resource_name,
    instances=TEST_INSTANCES,
)

print('gapic REST predict() on dedicated endpoint (patched):')
print(f'  Deployed model ID: {response.deployed_model_id}')
for pred in response.predictions:
    print(f'  {dict(pred)}')
````

### Patch 2: Unwrap HttpBody for `raw_predict`

The `$alt` patch alone isn't enough for `raw_predict`. Two independent issues remain:

**Request side:** The gapic REST transport serializes the `RawPredictRequest` protobuf using `MessageToJson`, wrapping the `HttpBody` as `{"httpBody": {"contentType": "...", "data": "<base64>"}}`. The dedicated DNS expects the raw body, not the protobuf JSON envelope.

**Response side:** The transport parses the raw JSON response (`{"predictions": [...]}`) with `json_format.Parse` into an `HttpBody` protobuf. Since `predictions` isn't an HttpBody field, it gets silently ignored and `response.data` comes back empty.

We fix both: unwrap the base64 request body, and re-wrap the raw response into a valid HttpBody JSON so `json_format.Parse` populates `data` correctly.

````python
import base64
from google.cloud.aiplatform_v1.services.prediction_service.transports.rest import (
    PredictionServiceRestTransport,
)

# --- Request fix: unwrap base64-encoded HttpBody.data ---
for cls_name in ('_BaseRawPredict', '_BaseStreamRawPredict'):
    inner_cls = getattr(_BasePredictionServiceRestTransport, cls_name, None)
    if inner_cls and hasattr(inner_cls, '_get_request_body_json'):
        original = inner_cls._get_request_body_json
        _patched_originals[f'{cls_name}._get_request_body_json'] = (inner_cls, original)

        def make_body_patch(orig):
            @staticmethod
            def patched(transcoded_request):
                body_json = orig(transcoded_request)
                body_dict = json.loads(body_json)
                # MessageToJson wraps HttpBody inside the request message:
                #   {"httpBody": {"contentType": "application/json", "data": "<base64>"}}
                http_body = body_dict.get('httpBody', body_dict)
                if 'data' in http_body:
                    return base64.b64decode(http_body['data']).decode('utf-8')
                return body_json
            return patched

        inner_cls._get_request_body_json = make_body_patch(original)
        print(f'Patched {cls_name}._get_request_body_json (request body unwrap)')

# --- Response fix: wrap raw JSON into HttpBody JSON for json_format.Parse ---
for cls_name in ('_RawPredict', '_StreamRawPredict'):
    concrete_cls = getattr(PredictionServiceRestTransport, cls_name, None)
    if concrete_cls and hasattr(concrete_cls, '_get_response'):
        original = concrete_cls._get_response
        _patched_originals[f'concrete.{cls_name}._get_response'] = (concrete_cls, original)

        def make_response_patch(orig):
            @staticmethod
            def patched(host, metadata, query_params, session, timeout, transcoded_request, body=None):
                response = orig(host, metadata, query_params, session, timeout, transcoded_request, body)
                if response.status_code < 400:
                    # The dedicated DNS returns raw JSON: {"predictions": [...]}
                    # But json_format.Parse expects HttpBody JSON: {"data": "<base64>", ...}
                    # Wrap the raw content so the protobuf parser populates HttpBody.data
                    ct = response.headers.get('Content-Type', 'application/json')
                    wrapped = json.dumps({
                        'contentType': ct,
                        'data': base64.b64encode(response.content).decode('ascii'),
                    })
                    response._content = wrapped.encode('utf-8')
                return response
            return patched

        concrete_cls._get_response = make_response_patch(original)
        print(f'Patched {cls_name}._get_response (response body wrap)')

# Test gapic raw_predict on dedicated endpoint
body = httpbody_pb2.HttpBody(
    data=json.dumps({'instances': TEST_INSTANCES}).encode('utf-8'),
    content_type='application/json',
)
response = predictor.raw_predict(
    endpoint=endpoint.resource_name,
    http_body=body,
)
predictions = json.loads(response.data)
print(f'\ngapic REST raw_predict() on dedicated endpoint (patched):')
for pred in predictions['predictions']:
    print(f'  {pred["label"]} ({pred["score"]})')
````

### Patch 3: Async gapic with `rest_asyncio`

The gapic async client (`PredictionServiceAsyncClient`) defaults to the `grpc_asyncio` transport, which hits the same `501` as sync gRPC. With our `$alt` patch applied, we can use the `rest_asyncio` transport instead — but it requires `StaticCredentials` from `google.auth.aio` (the standard OAuth2 credentials don't support the async auth flow).

### Patch 4: Fix `endpoint.predict_async()`

The SDK's `predict_async()` always routes through the regional gRPC endpoint. We fix it by replacing the cached internal gapic async client (`_prediction_async_client_value`) with our patched one that routes through the dedicated DNS.

The gRPC transport returns `501 MethodNotImplemented` when connecting to the dedicated DNS. The dedicated endpoint's DNS routes to an HTTP/REST serving frontend — it does not host a gRPC server. There is no client-side workaround.

### What We Learned

| Issue | Root Cause | Fix | Difficulty |
|-------|-----------|-----|-----------|
| gapic REST `predict()` → 400 | `$alt` query param | Patch `_get_query_params_json` | Moderate |
| gapic REST `raw_predict()` request → 500 | Request body base64-wrapped in `{"httpBody": {...}}` | Patch `_get_request_body_json` | Moderate |
| gapic REST `raw_predict()` response → empty | Response parsed as HttpBody, fields silently dropped | Patch `_get_response` to re-wrap | Moderate |
| gapic async → 400/501 | Wrong transport + `$alt` | `rest_asyncio` + `StaticCredentials` + patch | Hard |
| `predict_async()` → FailedPrecondition | Routes through regional API | Swap `_prediction_async_client_value` | Hack |
| gapic gRPC → 501 | Dedicated DNS isn't a gRPC server | **None — infrastructure limitation** | Impossible |

All the fixable issues boil down to: **the gapic REST transport was designed for the regional API, not the dedicated DNS.** The regional API expects the `$alt` parameter and protobuf JSON envelope; the dedicated DNS expects plain REST. The high-level SDK works because it bypasses gapic entirely for dedicated endpoints.
