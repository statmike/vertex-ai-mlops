# Serving Models on Cloud Run

Deploy the same custom prediction container to [Cloud Run](https://cloud.google.com/run/docs/overview/what-is-cloud-run) and compare with Vertex AI Endpoints. Same model, different platform — when to choose which.

**Story arc:**
1. Build the same FastAPI container used across the Serving notebooks
2. Deploy to Cloud Run (IAM-authenticated by default)
3. Authenticate correctly — Cloud Run requires **ID tokens**, not the access tokens used with Vertex AI
4. Deploy a second model revision, split traffic, and shift — Cloud Run's native canary pattern
5. Compare platforms: when Cloud Run, when Vertex AI

**Container portability:** The same Docker image runs on both Vertex AI Endpoints and Cloud Run. The `AIP_*` environment variables (`AIP_STORAGE_URI`, `AIP_PREDICT_ROUTE`, `AIP_HEALTH_ROUTE`, `AIP_HTTP_PORT`) are set by Vertex AI automatically and by us explicitly on Cloud Run. No code changes needed.

**Runtime:** ~10-15 minutes (Cloud Run deploys are fast — no GPU provisioning or model registry overhead).

**Prerequisites:** Familiarity with the [Dedicated Public Endpoint](./Vertex%20AI%20Dedicated%20Public%20Endpoint.ipynb) notebook pattern. For prediction method details, see [Prediction Methods](./Vertex%20AI%20Endpoint%20-%20Prediction%20Methods.ipynb). For autoscaling comparison, see [Autoscaling Deep Dive](./Vertex%20AI%20Endpoint%20-%20Autoscaling.ipynb).

## Download HuggingFace Models

Same two sentiment analysis models used in the [Dedicated Public Endpoint](./Vertex%20AI%20Dedicated%20Public%20Endpoint.ipynb) notebook. Two models enable the traffic splitting demo — their different label formats (`POSITIVE`/`NEGATIVE` vs `LABEL_0`/`LABEL_1`) confirm which revision is serving.

- **v1 (DistilBERT):** Smaller, faster — the initial deployment
- **v2 (BERT):** Larger, more accurate — the revision we swap to

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)
for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']
    results = classifier(instances)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
````

## Deploy to Cloud Run

Deploy the container as a Cloud Run service. Key differences from Vertex AI:

| Aspect | Vertex AI Endpoint | Cloud Run |
|--------|---|---|
| Model source | Model Registry (`artifact_uri`) | Container image (env var points to GCS) |
| Environment variables | Set automatically (`AIP_*`) | Set explicitly in service config |
| Scaling | Dedicated VMs, min 1 replica | Serverless, **scale-to-zero** by default |
| Deploy time | ~5-10 minutes | ~1-2 minutes |
| Auth | Access tokens (OAuth2) | **ID tokens (OIDC)** |

We set the same `AIP_*` environment variables that Vertex AI sets automatically — this is what makes the container portable between platforms.

````python
# Configure the Cloud Run service
parent = f'projects/{PROJECT_ID}/locations/{REGION}'
service_path = f'{parent}/services/{SERVICE_NAME}'

service = run_v2.Service(
    template=run_v2.RevisionTemplate(
        containers=[
            run_v2.Container(
                image=IMAGE_URI,
                ports=[run_v2.ContainerPort(container_port=8080)],
                env=[
                    run_v2.EnvVar(name='AIP_HTTP_PORT', value='8080'),
                    run_v2.EnvVar(name='AIP_HEALTH_ROUTE', value='/health'),
                    run_v2.EnvVar(name='AIP_PREDICT_ROUTE', value='/predict'),
                    run_v2.EnvVar(name='AIP_STORAGE_URI', value=MODELS['v1']['artifact_uri']),
                ],
                resources=run_v2.ResourceRequirements(
                    limits={'cpu': '4', 'memory': '8Gi'},
                ),
            )
        ],
        scaling=run_v2.RevisionScaling(
            min_instance_count=0,  # scale-to-zero (default)
            max_instance_count=5,
        ),
    ),
)

# Check for existing service
try:
    existing = cr.get_service(name=service_path)
    print(f'Service already exists: {existing.uri}')
    service_resp = existing
except Exception:
    print(f'Creating Cloud Run service: {SERVICE_NAME}')
    operation = cr.create_service(
        parent=parent,
        service=service,
        service_id=SERVICE_NAME,
    )
    service_resp = operation.result()
    print(f'Service created: {service_resp.uri}')

SERVICE_URL = service_resp.uri
print(f'\nService URL: {SERVICE_URL}')
print(f'Latest revision: {service_resp.latest_ready_revision.split("/")[-1]}')
````

## Authentication Deep Dive

Cloud Run services are **IAM-authenticated by default** — no `allUsers` binding means every request must prove its identity. But Cloud Run uses a different token type than Vertex AI.

### Access Tokens vs ID Tokens

| | Vertex AI Endpoints | Cloud Run |
|---|---|---|
| **Token type** | OAuth2 **access token** | OIDC **ID token** |
| **How to get it** | `credentials.token` after `credentials.refresh()` | `google.oauth2.id_token.fetch_id_token(request, audience)` |
| **What it proves** | "I have access to Google Cloud APIs" | "I am this identity, calling this specific service" |
| **Audience field** | Not applicable | Must match the Cloud Run service URL |
| **IAM permission** | `aiplatform.endpoints.predict` | `roles/run.invoker` |

The key difference: an **access token** is a general-purpose credential for Google Cloud APIs. An **ID token** is a service-specific credential that says "I am identity X, and I want to call service Y (the audience)."

> **Common mistake:** The [Understanding Prediction IO](./Understanding%20Prediction%20IO%20With%20FastAPI.ipynb) notebook uses `credentials.token` (an access token) to call Cloud Run — but it also adds `allUsers` to `roles/run.invoker`, making the service public. The access token is simply ignored. For production services, you want proper IAM auth with ID tokens.

### Grant Invoker Permission

Before we can call the service, our identity needs `roles/run.invoker`. Let's grant it to our current service account.

````python
# Get the current service account identity
SERVICE_ACCOUNT = !gcloud config list --format='value(core.account)'
SERVICE_ACCOUNT = SERVICE_ACCOUNT[0]
print(f'Service account: {SERVICE_ACCOUNT}')

# Grant roles/run.invoker on this specific service
policy = cr.get_iam_policy(request={'resource': service_path})

# Check if already granted
has_invoker = any(
    f'serviceAccount:{SERVICE_ACCOUNT}' in binding.members or
    f'user:{SERVICE_ACCOUNT}' in binding.members
    for binding in policy.bindings
    if binding.role == 'roles/run.invoker'
)

if has_invoker:
    print(f'Already has roles/run.invoker')
else:
    # Determine member type (service account vs user)
    if SERVICE_ACCOUNT.endswith('.iam.gserviceaccount.com'):
        member = f'serviceAccount:{SERVICE_ACCOUNT}'
    else:
        member = f'user:{SERVICE_ACCOUNT}'

    binding = policy_pb2.Binding(
        role='roles/run.invoker',
        members=[member],
    )
    policy.bindings.append(binding)
    cr.set_iam_policy(request={'resource': service_path, 'policy': policy})
    print(f'Granted roles/run.invoker to {member}')
    print('Waiting 30 seconds for IAM propagation...')
    time.sleep(30)
````

### The Wrong Way: Access Tokens

Access tokens are what you use for Vertex AI Endpoints. Let's see what happens when you use one with Cloud Run:

````python
# This is how you'd call a Vertex AI Endpoint — with an access token
credentials, _ = google.auth.default()
auth_req = google.auth.transport.requests.Request()
credentials.refresh(auth_req)

response = req_lib.post(
    f'{SERVICE_URL}/predict',
    headers={
        'Authorization': f'Bearer {credentials.token}',  # access token
        'Content-Type': 'application/json',
    },
    json={'instances': ['test sentence']},
    timeout=30,
)
print(f'Status: {response.status_code}')
if response.status_code != 200:
    print(f'Access tokens do not work for Cloud Run invocation.')
    print(f'Cloud Run requires ID tokens with the service URL as the audience.')
````

### The Right Way: ID Tokens

Cloud Run requires an OIDC **ID token** with the `audience` set to the service URL. The `google.oauth2.id_token` module handles this:

````python
# Get an ID token with the service URL as the audience
auth_req = google.auth.transport.requests.Request()
id_token = google.oauth2.id_token.fetch_id_token(auth_req, SERVICE_URL)

response = req_lib.post(
    f'{SERVICE_URL}/predict',
    headers={
        'Authorization': f'Bearer {id_token}',  # ID token — correct!
        'Content-Type': 'application/json',
    },
    json={'instances': ['This movie is wonderful!']},
    timeout=60,
)
print(f'Status: {response.status_code}')
if response.status_code == 200:
    print(f'Response: {response.json()}')
else:
    print(f'Error: {response.text[:200]}')
````

## Test Predictions

Now that auth is working, let's test with the same instances used across all Serving notebooks.

````python
TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]

def predict_cloud_run(instances, service_url=SERVICE_URL):
    """Send a prediction request to the Cloud Run service using ID token auth."""
    auth_req = google.auth.transport.requests.Request()
    id_token = google.oauth2.id_token.fetch_id_token(auth_req, service_url)

    response = req_lib.post(
        f'{service_url}/predict',
        headers={
            'Authorization': f'Bearer {id_token}',
            'Content-Type': 'application/json',
        },
        json={'instances': instances},
        timeout=60,
    )
    response.raise_for_status()
    return response.json()

def show_predictions(predictions):
    for instance, pred in zip(TEST_INSTANCES, predictions['predictions']):
        print(f'  "{instance[:60]}..."')
        print(f'    => {pred["label"]} (score: {pred["score"]})')
        print()

print('predict_cloud_run() and show_predictions() defined')
````

````python
print('DistilBERT (v1) predictions:\n')
response = predict_cloud_run(TEST_INSTANCES)
show_predictions(response)
````

## Public Access (Optional)

For services that don't need authentication (public APIs, webhooks), add `allUsers` to `roles/run.invoker`. Requests work without any auth header.

## Traffic Splitting Between Revisions

Cloud Run supports traffic splitting across **revisions** — each revision is a specific container image + configuration. When you update a service, Cloud Run creates a new revision automatically. You control how traffic is distributed across revisions.

This is Cloud Run's native blue/green and canary deployment pattern. Compare with Vertex AI Endpoints where you split traffic across **deployed models** (different model versions on the same endpoint).

**Plan:**
1. Note the current revision (DistilBERT, v1)
2. Update the service to use the BERT model (creates a new revision)
3. Split traffic 50/50 between old and new revisions
4. Verify both models are serving (different label formats confirm which)
5. Shift 100% to the new revision

### Current Revision

````python
# Get the current service and note the v1 revision name
service_resp = cr.get_service(name=service_path)
V1_REVISION = service_resp.latest_ready_revision.split('/')[-1]
print(f'Current (v1) revision: {V1_REVISION}')
print(f'Model: DistilBERT (labels: POSITIVE / NEGATIVE)')
````

### Deploy v2 Revision with Traffic Split

Update the service to use the BERT model. By setting the `traffic` field in the same update, we can split traffic between the old and new revisions in a single operation.

````python
# Get the current service to modify it
service = cr.get_service(name=service_path)

# Update the model URI to v2 (BERT) — this creates a new revision
for env_var in service.template.containers[0].env:
    if env_var.name == 'AIP_STORAGE_URI':
        env_var.value = MODELS['v2']['artifact_uri']
        break

# Split traffic: 50% to v1 revision, 50% to the new (latest) revision
service.traffic = [
    run_v2.TrafficTarget(
        type_=run_v2.TrafficTargetAllocationType.TRAFFIC_TARGET_ALLOCATION_TYPE_REVISION,
        revision=V1_REVISION,
        percent=50,
    ),
    run_v2.TrafficTarget(
        type_=run_v2.TrafficTargetAllocationType.TRAFFIC_TARGET_ALLOCATION_TYPE_LATEST,
        percent=50,
    ),
]

print('Updating service with v2 model and 50/50 traffic split...')
operation = cr.update_service(service=service)
service_resp = operation.result()

V2_REVISION = service_resp.latest_ready_revision.split('/')[-1]
print(f'\nv1 revision: {V1_REVISION} (DistilBERT)')
print(f'v2 revision: {V2_REVISION} (BERT)')
print(f'\nTraffic:')
for target in service_resp.traffic:
    rev = target.revision.split('/')[-1] if target.revision else 'LATEST'
    print(f'  {rev}: {target.percent}%')
````

### Verify Both Models Are Serving

Send multiple requests. With 50/50 traffic, roughly half should return DistilBERT labels (`POSITIVE`/`NEGATIVE`) and half should return BERT labels (`LABEL_0`/`LABEL_1`).

````python
# Send 20 requests and count which model serves each
from collections import Counter

print('Sending 20 requests to check traffic distribution...\n')
label_counts = Counter()

for i in range(20):
    result = predict_cloud_run(['This movie is great!'])
    label = result['predictions'][0]['label']
    label_counts[label] += 1

print('Results:')
for label, count in sorted(label_counts.items()):
    model = 'DistilBERT (v1)' if label in ('POSITIVE', 'NEGATIVE') else 'BERT (v2)'
    print(f'  {label:>10s}: {count:2d} requests — {model}')

distilbert = sum(c for l, c in label_counts.items() if l in ('POSITIVE', 'NEGATIVE'))
bert = sum(c for l, c in label_counts.items() if l not in ('POSITIVE', 'NEGATIVE'))
print(f'\nDistribution: DistilBERT {distilbert}/20, BERT {bert}/20')
````

### Shift 100% Traffic to v2 (BERT)

````python
# Route all traffic to the latest (v2) revision
service = cr.get_service(name=service_path)
service.traffic = [
    run_v2.TrafficTarget(
        type_=run_v2.TrafficTargetAllocationType.TRAFFIC_TARGET_ALLOCATION_TYPE_LATEST,
        percent=100,
    ),
]

operation = cr.update_service(service=service)
service_resp = operation.result()

print('Traffic shifted to 100% v2 (BERT)')
print(f'\nTraffic:')
for target in service_resp.traffic:
    rev = target.revision.split('/')[-1] if target.revision else 'LATEST'
    print(f'  {rev}: {target.percent}%')
````

## Configuration Reference

### Cloud Run Autoscaling

Cloud Run scales based on **concurrent requests per instance** (default: 80). This is fundamentally different from Vertex AI, which scales on CPU/GPU utilization or request rate.

| Feature | Cloud Run | Vertex AI Endpoint |
|---------|-----------|-------------------|
| **Scale-to-zero** | Yes (default) | Preview only, drops requests (429) |
| **Min instances** | 0 (default) | 1 (default) |
| **Max instances** | 1000 (default) | Set at deploy time |
| **Scaling metric** | Concurrent requests per instance | CPU, GPU duty cycle, or request count |
| **Concurrency** | Configurable (1-1000 per instance) | 1 request per replica |
| **Cold start** | Container startup time | Not applicable (always running) |
| **CPU allocation** | Per-request or always-on | Always allocated |

```python
# Example: Configure autoscaling
service.template.scaling = run_v2.RevisionScaling(
    min_instance_count=1,    # keep 1 warm (avoid cold starts)
    max_instance_count=10,   # scale up to 10 instances
)
service.template.max_instance_request_concurrency = 4  # 4 concurrent requests per instance
```

### GPU on Cloud Run

Cloud Run supports GPU inference with NVIDIA L4 GPUs. Configuration:

```python
service.template.containers[0].resources = run_v2.ResourceRequirements(
    limits={'cpu': '8', 'memory': '32Gi', 'nvidia.com/gpu': '1'},
)
service.template.node_selector = {'run.googleapis.com/accelerator': 'nvidia-l4'}
```

**Constraints:**
- Available regions: `us-central1`, `europe-west1`, `asia-southeast1` (check current availability)
- Requires `min_instance_count >= 1` (no scale-to-zero with GPU)
- `max_instance_request_concurrency` must be 1-4 for GPU instances
- Supported GPUs: NVIDIA L4 (most common), A100 (limited availability)

Not demonstrated in this notebook because our small sentiment model runs well on CPU.

### When to Choose Cloud Run vs Vertex AI

| Need | Cloud Run | Vertex AI |
|------|:-:|:-:|
| **Simple HTTP serving** | Best | Good |
| **Scale-to-zero** | Yes | Limited |
| **Fast deploys** | ~1-2 min | ~5-10 min |
| **Model Registry + versioning** | No | Yes |
| **Batch prediction** | No | Yes |
| **Model monitoring** | No | Yes |
| **Explainability / feature attributions** | No | Yes |
| **Traffic splitting** | Revision-based | Model-based |
| **GPU inference** | L4 | Full menu (T4, V100, A100, L4, H100) |
| **Auth model** | IAM + ID tokens | IAM + access tokens |
| **Max payload** | 32 MB | 10 MB (dedicated) / 1.5 MB (shared) |
| **Max timeout** | 60 min | 60 min (dedicated) / 60s (shared) |
| **gRPC** | Yes | Yes (dedicated/PSC only) |
| **Streaming (SSE)** | Yes | Yes (dedicated/PSC only) |
| **Cost model** | Per-request (or always-on) | Per-replica-hour |
| **Container portability** | Any container | Any container (with AIP_* env vars) |

**Rule of thumb:**
- **Choose Cloud Run** when you want simplicity, scale-to-zero, and don't need ML-specific features (monitoring, explainability, batch prediction, model registry)
- **Choose Vertex AI** when you need the ML platform features — model versioning, A/B testing with model-level traffic splits, batch prediction, monitoring, or explainability
- **Either works** for GPU inference, traffic splitting, and high-throughput serving — the trade-off is platform features vs operational simplicity
