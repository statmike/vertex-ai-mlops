# Vertex AI Pre-built Serving Containers

## Overview

Deploy models without writing containers. Vertex AI provides **pre-built serving containers** for TensorFlow, scikit-learn, XGBoost, and PyTorch. You supply the model artifacts in the expected format, point to a pre-built container URI, and Vertex AI handles the rest.

**Contrast with other notebooks in this series:** every other notebook in the MLOps/Serving series builds a custom FastAPI container, pushes it to Artifact Registry via Cloud Build, and deploys it. This notebook shows the simpler path — when you do NOT need a custom container at all.

### What you will learn

- Model export formats: TorchScript/MAR archives for PyTorch, SavedModel for TensorFlow
- Pre-built container URIs and how to find them
- Uploading a model to the Vertex AI Model Registry with a pre-built container
- Deploying and getting predictions — same endpoint interface, zero container code

### When to use pre-built containers

- Your model is in a standard framework format (SavedModel, TorchScript, pickle, etc.)
- You do not need custom pre/post-processing logic beyond what the serving framework supports
- You want the fastest path from trained model to deployed endpoint

## Download HuggingFace Models

- **v1 (DistilBERT):** Smaller, faster — good for initial deployment
- **v2 (BERT):** Larger, more accurate — the model we'll swap to

```python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
```

```python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        # upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    # store the GCS URI
    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
```

## Pre-built Container Overview

Vertex AI provides pre-built serving containers for the most popular ML frameworks. You do **not** need a Dockerfile, Cloud Build, or Artifact Registry setup — just upload your model artifacts to GCS and reference the container URI when registering the model.

| Framework | Container URI | Model Format |
|-----------|---------------|-------------|
| TensorFlow | `us-docker.pkg.dev/vertex-ai/prediction/tf2-cpu.2-15:latest` | SavedModel |
| scikit-learn | `us-docker.pkg.dev/vertex-ai/prediction/sklearn-cpu.1-3:latest` | pickle / joblib |
| XGBoost | `us-docker.pkg.dev/vertex-ai/prediction/xgboost-cpu.2-1:latest` | JSON / binary |
| PyTorch | `us-docker.pkg.dev/vertex-ai/prediction/pytorch-cpu.2-4:latest` | TorchScript (MAR archive) |

**How it works:**

1. Export your model in the framework-specific format
2. Upload model artifacts to a GCS path
3. Call `aiplatform.Model.upload()` with the pre-built container URI
4. Deploy to an endpoint — the container handles `/predict` and `/health` automatically

```python
# PyTorch pre-built container URI
PYTORCH_CONTAINER = 'us-docker.pkg.dev/vertex-ai/prediction/pytorch-cpu.2-4:latest'

# TensorFlow pre-built container URI
TF_CONTAINER = 'us-docker.pkg.dev/vertex-ai/prediction/tf2-cpu.2-15:latest'

print(f'PyTorch container: {PYTORCH_CONTAINER}')
print(f'TensorFlow container: {TF_CONTAINER}')
```

## Approach 1: PyTorch Pre-built Container (TorchServe)

Export DistilBERT as a TorchServe Model Archive (`.mar`), upload to GCS, and deploy with the PyTorch pre-built container.

The PyTorch pre-built container runs **TorchServe** under the hood. It expects model artifacts packaged as a `.mar` archive using `torch-model-archiver`.

### Create a TorchServe Handler

TorchServe uses a handler class to define how input is preprocessed, how inference is run, and how output is postprocessed. This is the only custom code needed — no Dockerfile, no FastAPI.

```python
HANDLER_CODE = '''
import json
import torch
import inspect
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from ts.torch_handler.base_handler import BaseHandler

class SentimentHandler(BaseHandler):
    def initialize(self, context):
        self.manifest = context.manifest
        model_dir = context.system_properties.get("model_dir")
        self.model = AutoModelForSequenceClassification.from_pretrained(model_dir)
        self.tokenizer = AutoTokenizer.from_pretrained(model_dir)
        self.model.eval()
        self.id2label = self.model.config.id2label
        # Get the model's accepted forward() arguments so we can filter tokenizer output
        self.forward_params = set(inspect.signature(self.model.forward).parameters.keys())

    def preprocess(self, data):
        texts = []
        for row in data:
            body = row.get("body") or row.get("data")
            if isinstance(body, (bytes, bytearray)):
                body = body.decode("utf-8")
            if isinstance(body, str):
                body = json.loads(body)
            text = body.get("text", body.get("instances", [{}])[0].get("text", ""))
            texts.append(text)
        encoded = self.tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=128)
        # Only pass keys the model accepts (e.g. DistilBERT does not use token_type_ids)
        return {k: v for k, v in encoded.items() if k in self.forward_params}

    def inference(self, inputs):
        with torch.no_grad():
            outputs = self.model(**inputs)
        return outputs.logits

    def postprocess(self, outputs):
        probs = torch.nn.functional.softmax(outputs, dim=1)
        results = []
        for prob in probs:
            idx = prob.argmax().item()
            results.append({"label": self.id2label[idx], "score": prob[idx].item()})
        return results
'''

handler_path = f'{DIR}/handler.py'
with open(handler_path, 'w') as f:
    f.write(HANDLER_CODE)
print(f'Wrote handler: {handler_path}')
```

### Create Model Archive (.mar)

```python
# Save model artifacts locally
model_artifacts_dir = f'{DIR}/model_store/distilbert'
os.makedirs(model_artifacts_dir, exist_ok=True)

from transformers import AutoModelForSequenceClassification, AutoTokenizer
model = AutoModelForSequenceClassification.from_pretrained(MODELS['v1']['hub_name'])
tokenizer = AutoTokenizer.from_pretrained(MODELS['v1']['hub_name'])
model.save_pretrained(model_artifacts_dir)
tokenizer.save_pretrained(model_artifacts_dir)

# Package as MAR using torch-model-archiver
!pip install -q torch-model-archiver

# Vertex AI pre-built PyTorch container expects exactly "model.mar"
!torch-model-archiver \
    --model-name model \
    --version 1.0 \
    --handler {handler_path} \
    --extra-files {model_artifacts_dir} \
    --export-path {DIR}/model_store \
    --force

print(f'Model archive created: {DIR}/model_store/model.mar')
```

```python
# Clean up any old MAR files and upload the correctly named one
mar_path = f'{DIR}/model_store/model.mar'
gcs_mar_prefix = f'{SERIES}/{EXPERIMENT}/pytorch'

# Remove stale files from prior run
for blob in bucket.list_blobs(prefix=gcs_mar_prefix):
    blob.delete()
    print(f'Deleted old: gs://{GCS_BUCKET}/{blob.name}')

blob = bucket.blob(f'{gcs_mar_prefix}/model.mar')
blob.upload_from_filename(mar_path)
print(f'Uploaded to gs://{GCS_BUCKET}/{gcs_mar_prefix}/model.mar')
```

```python
# Clean up any model/endpoint from a prior failed attempt
for ep in aiplatform.Endpoint.list(filter=f'display_name="{SERIES}-{EXPERIMENT}-pytorch-endpoint"'):
    ep.undeploy_all()
    ep.delete()
    print(f'Cleaned up old endpoint: {ep.resource_name}')
for m in aiplatform.Model.list(filter=f'display_name="{SERIES}-{EXPERIMENT}-pytorch"'):
    m.delete()
    print(f'Cleaned up old model: {m.resource_name}')

# Upload to Vertex AI Model Registry with pre-built PyTorch container
model_pt = aiplatform.Model.upload(
    display_name=f'{SERIES}-{EXPERIMENT}-pytorch',
    artifact_uri=f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/pytorch/',
    serving_container_image_uri=PYTORCH_CONTAINER,
)
model_pt.wait()
print(f'Model uploaded: {model_pt.resource_name}')
```

```python
# Deploy to an endpoint
endpoint_pt = aiplatform.Endpoint.create(
    display_name=f'{SERIES}-{EXPERIMENT}-pytorch-endpoint',
)

model_pt.deploy(
    endpoint=endpoint_pt,
    deployed_model_display_name=f'{SERIES}-{EXPERIMENT}-pytorch-deployed',
    machine_type='n1-standard-4',
    min_replica_count=1,
    max_replica_count=1,
)
print(f'Deployed to: {endpoint_pt.resource_name}')
```

```python
# Test predictions
test_instances = [
    {"data": json.dumps({"text": "This movie is absolutely wonderful and I loved every minute of it!"})},
    {"data": json.dumps({"text": "The food was terrible and the service was even worse."})},
    {"data": json.dumps({"text": "An average experience, nothing special but not bad either."})},
]

response_pt = endpoint_pt.predict(instances=test_instances)

print('PyTorch Pre-built Container Predictions:')
print('-' * 50)
for instance, pred in zip(test_instances, response_pt.predictions):
    text = json.loads(instance['data'])['text']
    print(f'  "{text[:60]}..."')
    print(f'    => {pred}')
    print()
```

## Approach 2: TensorFlow Pre-built Container (SavedModel)

Deploy a Keras model using the TF Serving pre-built container. The container runs **TensorFlow Serving** and expects a SavedModel directory structure with a numeric version subdirectory (e.g., `1/`).

> **Note:** HuggingFace Transformers v5+ removed TensorFlow model support (`TFAutoModel*` classes). For TF pre-built containers, use native Keras/TF models. This section trains a simple IMDB sentiment classifier in Keras to demonstrate the pattern — same deployment workflow, different model framework.

```python
import tensorflow as tf
import numpy as np

# Load IMDB dataset (built into Keras — pre-tokenized as integer sequences)
VOCAB_SIZE = 10000
MAX_LEN = 256

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=VOCAB_SIZE)
x_train = tf.keras.utils.pad_sequences(x_train, maxlen=MAX_LEN)
x_test = tf.keras.utils.pad_sequences(x_test, maxlen=MAX_LEN)

# Simple text classification model
tf_model = tf.keras.Sequential([
    tf.keras.layers.Embedding(VOCAB_SIZE, 32, input_length=MAX_LEN),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax'),
])
tf_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
tf_model.fit(x_train, y_train, epochs=5, batch_size=512, validation_split=0.2, verbose=1)

loss, acc = tf_model.evaluate(x_test, y_test, verbose=0)
print(f'\nTest accuracy: {acc:.4f}')

# TF Serving expects a versioned subdirectory with SavedModel format
# Keras 3 uses model.export() for SavedModel (model.save() requires .keras/.h5)
tf_export_path = f'{DIR}/tf_saved_model/1'
tf_model.export(tf_export_path, format='tf_saved_model')
print(f'SavedModel exported to: {tf_export_path}')
```

```python
# Upload the SavedModel directory to GCS
tf_gcs_prefix = f'{SERIES}/{EXPERIMENT}/tensorflow'

local_base = f'{DIR}/tf_saved_model'
for root, dirs, files in os.walk(local_base):
    for fname in files:
        local_path = os.path.join(root, fname)
        blob_name = f'{tf_gcs_prefix}/{os.path.relpath(local_path, local_base)}'
        bucket.blob(blob_name).upload_from_filename(local_path)

print(f'Uploaded SavedModel to gs://{GCS_BUCKET}/{tf_gcs_prefix}/')
```

```python
# Clean up any model/endpoint from a prior failed attempt
for ep in aiplatform.Endpoint.list(filter=f'display_name="{SERIES}-{EXPERIMENT}-tensorflow-endpoint"'):
    ep.undeploy_all()
    ep.delete()
    print(f'Cleaned up old endpoint: {ep.resource_name}')
for m in aiplatform.Model.list(filter=f'display_name="{SERIES}-{EXPERIMENT}-tensorflow"'):
    m.delete()
    print(f'Cleaned up old model: {m.resource_name}')

# Upload to Vertex AI Model Registry with pre-built TF container
# artifact_uri must point to the directory containing saved_model.pb directly
model_tf = aiplatform.Model.upload(
    display_name=f'{SERIES}-{EXPERIMENT}-tensorflow',
    artifact_uri=f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/tensorflow/1/',
    serving_container_image_uri=TF_CONTAINER,
)
model_tf.wait()
print(f'Model uploaded: {model_tf.resource_name}')
```

```python
# Deploy to an endpoint
endpoint_tf = aiplatform.Endpoint.create(
    display_name=f'{SERIES}-{EXPERIMENT}-tensorflow-endpoint',
)

model_tf.deploy(
    endpoint=endpoint_tf,
    deployed_model_display_name=f'{SERIES}-{EXPERIMENT}-tensorflow-deployed',
    machine_type='n1-standard-4',
    min_replica_count=1,
    max_replica_count=1,
)
print(f'Deployed to: {endpoint_tf.resource_name}')
```

```python
# Test predictions using IMDB test samples
# TF Serving expects the input format matching the model's serving signature
test_indices = [0, 1, 2]
test_samples = x_test[test_indices].tolist()
test_labels = ['NEGATIVE' if y == 0 else 'POSITIVE' for y in y_test[test_indices]]

# Get the IMDB word index to decode reviews for display
word_index = tf.keras.datasets.imdb.get_word_index()
reverse_index = {v + 3: k for k, v in word_index.items()}
reverse_index[0], reverse_index[1], reverse_index[2] = '<pad>', '<start>', '<unk>'

def decode_review(encoded):
    return ' '.join(reverse_index.get(i, '?') for i in encoded if i != 0)

# Get the input key name from the exported SavedModel signature
# (Keras names it keras_tensor, keras_tensor_1, etc. depending on kernel state)
loaded = tf.saved_model.load(f'{DIR}/tf_saved_model/1')
input_key = list(loaded.signatures['serving_default'].structured_input_signature[1].keys())[0]
print(f'Serving signature input key: {input_key}')

instances = [{input_key: sample} for sample in test_samples]
response_tf = endpoint_tf.predict(instances=instances)

print('\nTensorFlow Pre-built Container Predictions (IMDB):')
print('-' * 50)
for idx, pred in zip(test_indices, response_tf.predictions):
    review_text = decode_review(x_test[idx])
    # TF Serving may return probs as a raw list or a dict with output key
    probs = pred if isinstance(pred, list) else pred.get('output_0', list(pred.values())[0])
    label = 'POSITIVE' if probs[1] > probs[0] else 'NEGATIVE'
    score = max(probs)
    print(f'  Review: "{review_text[:80]}..."')
    print(f'    => {label} (score: {score:.4f}) [actual: {test_labels[test_indices.index(idx)]}]')
    print()
```

## Comparison: Pre-built vs Custom Containers

| Aspect | Pre-built Container | Custom Container |
|--------|-------------------|------------------|
| **Container build** | None | Dockerfile + Cloud Build |
| **Serving framework** | TorchServe / TF Serving | Any (FastAPI, Flask, etc.) |
| **Model format** | Framework-specific (SavedModel, MAR, pickle) | Any format you can load |
| **Custom pre/post processing** | Limited (TorchServe handler) | Full control |
| **Multi-model** | One model per container | Multiple models possible |
| **I/O format** | Framework default | Fully custom |
| **Setup time** | Minutes | Hours |
| **Flexibility** | Low | High |
| **Best for** | Standard models, quick deployment | Complex serving logic |

### Decision Framework

**Use pre-built containers when:**
- Your model is in a standard format (SavedModel, TorchScript, pickle, joblib)
- You need the fastest path from trained model to deployed endpoint
- The default serving behavior (tokenization, inference, output) is sufficient
- You are prototyping or deploying standard model architectures

**Use custom containers (like the other notebooks in this series) when:**
- You need custom pre/post-processing (e.g., text tokenization before inference)
- You want to serve multiple models from a single endpoint
- You need a custom health check or monitoring logic
- Your model format is not supported by pre-built containers
- You need to control the HTTP interface (custom routes, batch handling, etc.)

> **Note:** The PyTorch pre-built container with a TorchServe handler (Approach 1 above) blurs this line — it lets you add custom pre/post-processing via the handler while still avoiding Dockerfile and Cloud Build. For many NLP models, this is a good middle ground.
