# Vertex AI PSC Endpoint - Pipeline Model Swap

Automate model rollout on a [Private Service Connect (PSC)](https://cloud.google.com/vertex-ai/docs/predictions/private-service-connect) endpoint using [Vertex AI Pipelines](https://cloud.google.com/vertex-ai/docs/pipelines/introduction). This notebook builds on the manual PSC workflow in [Vertex AI Private Endpoint With PSC](./Vertex%20AI%20Private%20Endpoint%20With%20PSC.ipynb) and replaces the manual model swap with a Kubeflow Pipeline.

**Story arc:**
1. Set up PSC infrastructure manually (endpoint, deploy v1, networking, VM)
2. Verify DistilBERT is serving via the private IP
3. Define a [Kubeflow Pipeline](https://www.kubeflow.org/docs/components/pipelines/) that automates model rollout with automatic rollback
4. Run the pipeline to deploy BERT, shift traffic, verify the deployment, and undeploy the old model
5. Independently verify from the VM that BERT is now serving via PSC

**Why a pipeline?** The manual swap in the PSC notebook works for learning, but in production you want a repeatable, auditable process. The pipeline handles: pre-deployment validation, zero-traffic deploy, traffic shift with propagation wait, deployment verification, automatic rollback on failure, old model cleanup, and status notification. One pipeline run, one click.

**Testing approach:** PSC forwarding rules are local to the VPC — they are [not transitive through VPC peering](https://cloud.google.com/vpc/docs/vpc-peering#psc-forwarding-rules). Since Vertex AI Pipelines workers connect via peering, they cannot reach the PSC IP directly. The pipeline validates the deployment via the Vertex AI management API (confirming the model is deployed and receiving 100% traffic), and the GCE VM provides independent prediction testing through the actual PSC private IP after the pipeline completes.

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params) — manual deploy
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params) — pipeline deploy

## Download HuggingFace Models

Download two sentiment analysis models from HuggingFace and upload them to GCS. Both models are fine-tuned on the [SST-2](https://huggingface.co/datasets/stanfordnlp/sst2) dataset for binary sentiment classification (POSITIVE/NEGATIVE).

- **v1 (DistilBERT):** Smaller, faster — deployed manually as the initial model
- **v2 (BERT):** Larger, more accurate — deployed by the pipeline

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'
    
    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue
    
    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)
        
        # upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    # store the GCS URI
    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

````python
for version, info in MODELS.items():
    blobs = list(bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/models/{version}'))
    print(f'{version} ({info["display_name"]}): {len(blobs)} files at {info["artifact_uri"]}')
````

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)
for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']
    results = classifier(instances)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
````

Create a small GCE VM in the same VPC to independently verify predictions through the PSC private IP — both before and after the pipeline runs. Since PSC forwarding rules are [not transitive through VPC peering](https://cloud.google.com/vpc/docs/vpc-peering#psc-forwarding-rules), the pipeline validates deployment health via the management API while the VM provides the end-to-end PSC path validation.

````python
print('Waiting for VM to be ready ...')
time.sleep(30)
!gcloud compute instances describe {VM_NAME} --zone={REGION}-a --project={PROJECT_ID} --format='value(status)'
````

````python
def predict_via_vm(instances, endpoint_id=None, psc_ip=None):
    """Send a prediction request to the PSC private endpoint via the GCE VM."""
    if psc_ip is None:
        psc_ip = PSC_IP
    if endpoint_id is None:
        endpoint_id = endpoint.name.split('/')[-1]
    
    payload = json.dumps({'instances': instances})
    
    curl_cmd = (
        f'curl -s -X POST '
        f'-H "Content-Type: application/json" '
        f'-H "Authorization: Bearer $(gcloud auth print-access-token)" '
        f"-d '{payload}' "
        f'https://{psc_ip}/v1/projects/{PROJECT_ID}/locations/{REGION}/endpoints/{endpoint_id}:predict '
        f'-k'
    )
    
    result = subprocess.run(
        [
            'gcloud', 'compute', 'ssh', VM_NAME,
            f'--zone={REGION}-a',
            f'--project={PROJECT_ID}',
            '--command', curl_cmd,
            '--quiet',
        ],
        capture_output=True, text=True,
    )
    
    if result.stdout.strip():
        try:
            return json.loads(result.stdout)
        except json.JSONDecodeError:
            print(f'Unexpected response: {result.stdout.strip()}')
            return None
    
    print(f'No response from endpoint (exit code {result.returncode})')
    if result.stderr.strip():
        lines = [l for l in result.stderr.strip().split('\n')
                 if 'WARNING:' not in l and 'known hosts' not in l]
        if lines:
            print('\n'.join(lines))
    return None

# Test sentences for sentiment analysis
TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]
````

## Define Pipeline: Automated Model Swap

Instead of manually deploying v2, shifting traffic, and undeploying v1 (as done in the [PSC notebook](./Vertex%20AI%20Private%20Endpoint%20With%20PSC.ipynb)), define a [Kubeflow Pipeline](https://www.kubeflow.org/docs/components/pipelines/) that automates the entire rollout. Each step is a KFP v2 component that runs on Vertex AI Pipelines.

**Pipeline flow:**
```
ExitHandler(notify_status)
  verify_endpoint ─> deploy_model (0%) ─> shift_traffic (100%) ─> verify_deployment
                                                                        │
                                                             ┌──────────┴──────────┐
                                                        If verified:          Else (failed):
                                                      undeploy_old_models    rollback_traffic
```

> **Why not test predictions from the pipeline?** PSC forwarding rules are local to the VPC and [not accessible from peered networks](https://cloud.google.com/vpc/docs/vpc-peering#psc-forwarding-rules). Since Vertex AI Pipelines workers run on a peered VPC (via the `network` parameter), they cannot reach the PSC private IP. The pipeline validates deployment health through the management API, and the GCE VM validates predictions through the PSC path independently.

### Pipeline Components

````python
@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-aiplatform']
)
def verify_endpoint(
    project: str,
    region: str,
    endpoint_resource_name: str,
) -> str:
    """Verify endpoint exists and list current deployments."""
    from google.cloud import aiplatform
    import json
    
    aiplatform.init(project=project, location=region)
    endpoint = aiplatform.PrivateEndpoint(endpoint_resource_name)
    
    deployments = []
    for model in endpoint.list_models():
        traffic = endpoint.traffic_split.get(model.id, 0)
        deployments.append({
            'id': model.id,
            'display_name': model.display_name,
            'traffic_pct': traffic,
        })
        print(f'  Deployed: {model.display_name} (id={model.id}, traffic={traffic}%)')
    
    print(f'\nEndpoint {endpoint.display_name} verified with {len(deployments)} deployed model(s)')
    return json.dumps(deployments)
````

````python
@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-aiplatform']
)
def deploy_model(
    project: str,
    region: str,
    endpoint_resource_name: str,
    model_resource_name: str,
    deployed_model_display_name: str,
    machine_type: str,
    min_replicas: int,
    max_replicas: int,
) -> str:
    """Deploy model at 0% traffic. Returns the deployed model display name."""
    from google.cloud import aiplatform
    
    aiplatform.init(project=project, location=region)
    endpoint = aiplatform.PrivateEndpoint(endpoint_resource_name)
    model = aiplatform.Model(model_resource_name)
    
    # Check if already deployed
    for deployed in endpoint.list_models():
        if deployed.display_name == deployed_model_display_name:
            print(f'{deployed_model_display_name} is already deployed (id={deployed.id}), skipping deploy')
            return deployed_model_display_name
    
    # Deploy at 0% traffic
    print(f'Deploying {model.display_name} as {deployed_model_display_name} at 0% traffic ...')
    endpoint.deploy(
        model=model,
        deployed_model_display_name=deployed_model_display_name,
        machine_type=machine_type,
        min_replica_count=min_replicas,
        max_replica_count=max_replicas,
        traffic_percentage=0,
    )
    print(f'Deploy complete. Traffic split: {endpoint.traffic_split}')
    return deployed_model_display_name
````

````python
@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-aiplatform']
)
def shift_traffic(
    project: str,
    region: str,
    endpoint_resource_name: str,
    target_display_name: str,
    wait_seconds: int,
) -> str:
    """Shift 100% traffic to the target model, 0% to all others. Wait for propagation."""
    from google.cloud import aiplatform
    import time
    
    aiplatform.init(project=project, location=region)
    endpoint = aiplatform.PrivateEndpoint(endpoint_resource_name)
    
    old_split = dict(endpoint.traffic_split)
    print(f'Current traffic split: {old_split}')
    
    # Build new split: 100% to target, 0% to everything else
    new_split = {}
    target_found = False
    for deployed in endpoint.list_models():
        if deployed.display_name == target_display_name:
            new_split[deployed.id] = 100
            target_found = True
        else:
            new_split[deployed.id] = 0
    
    if not target_found:
        raise RuntimeError(f'Model {target_display_name} not found on endpoint')
    
    endpoint.update(traffic_split=new_split)
    print(f'New traffic split: {new_split}')
    
    # Wait for routing to propagate (dedicated/PSC endpoints need this)
    print(f'Waiting {wait_seconds}s for traffic routing to propagate ...')
    time.sleep(wait_seconds)
    
    return target_display_name
````

````python
@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-aiplatform']
)
def verify_deployment(
    project: str,
    region: str,
    endpoint_resource_name: str,
    target_display_name: str,
) -> bool:
    """Verify the target model is deployed and receiving 100% traffic.
    
    PSC forwarding rules are not transitive through VPC peering, so pipeline
    workers cannot reach the PSC IP directly. This component validates the
    deployment via the management API instead. Prediction testing through the
    actual PSC path is done independently from the GCE VM after the pipeline.
    """
    from google.cloud import aiplatform
    
    aiplatform.init(project=project, location=region)
    endpoint = aiplatform.PrivateEndpoint(endpoint_resource_name)
    
    target_id = None
    for deployed in endpoint.list_models():
        traffic = endpoint.traffic_split.get(deployed.id, 0)
        print(f'  {deployed.display_name}: id={deployed.id}, traffic={traffic}%')
        if deployed.display_name == target_display_name:
            target_id = deployed.id
    
    if target_id is None:
        print(f'\nFAIL: {target_display_name} not found on endpoint')
        return False
    
    target_traffic = endpoint.traffic_split.get(target_id, 0)
    if target_traffic != 100:
        print(f'\nFAIL: {target_display_name} has {target_traffic}% traffic, expected 100%')
        return False
    
    print(f'\nVerified: {target_display_name} is deployed with 100% traffic')
    return True
````

````python
@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-aiplatform']
)
def undeploy_old_models(
    project: str,
    region: str,
    endpoint_resource_name: str,
    keep_display_name: str,
) -> str:
    """Undeploy all models except the one with keep_display_name."""
    from google.cloud import aiplatform
    import json
    
    aiplatform.init(project=project, location=region)
    endpoint = aiplatform.PrivateEndpoint(endpoint_resource_name)
    
    removed = []
    for deployed in endpoint.list_models():
        if deployed.display_name != keep_display_name:
            print(f'Undeploying {deployed.display_name} (id={deployed.id}) ...')
            endpoint.undeploy(deployed_model_id=deployed.id)
            removed.append(deployed.display_name)
    
    print(f'\nRemaining deployed models:')
    for m in endpoint.list_models():
        print(f'  {m.display_name} — traffic: {endpoint.traffic_split.get(m.id, 0)}%')
    
    return json.dumps(removed)
````

````python
@dsl.component(
    base_image='python:3.10',
    packages_to_install=['google-cloud-aiplatform']
)
def rollback_traffic(
    project: str,
    region: str,
    endpoint_resource_name: str,
    failed_display_name: str,
    wait_seconds: int,
) -> str:
    """Roll back traffic: shift 100% to the previous model, 0% to the failed one."""
    from google.cloud import aiplatform
    import time

    aiplatform.init(project=project, location=region)
    endpoint = aiplatform.PrivateEndpoint(endpoint_resource_name)

    new_split = {}
    rollback_target = None
    for deployed in endpoint.list_models():
        if deployed.display_name == failed_display_name:
            new_split[deployed.id] = 0
        else:
            new_split[deployed.id] = 100
            rollback_target = deployed.display_name

    if rollback_target is None:
        raise RuntimeError('No model to roll back to — only the failed model is deployed')

    print(f'Rolling back: {failed_display_name} -> 0%, {rollback_target} -> 100%')
    endpoint.update(traffic_split=new_split)

    print(f'Waiting {wait_seconds}s for traffic routing to propagate ...')
    time.sleep(wait_seconds)

    print(f'Rollback complete. Traffic restored to {rollback_target}')
    return rollback_target
````

````python
@dsl.component(
    base_image='python:3.10',
)
def notify_status(
    status: dsl.PipelineTaskFinalStatus,
    notification_email: str,
) -> str:
    """Log pipeline completion status. In production, integrate with email/Pub/Sub."""
    import json
    
    result = {
        'state': status.state,
        'pipeline_job': status.pipeline_job_resource_name,
        'error': getattr(status, 'error_message', None),
    }
    
    if status.state == 'SUCCEEDED':
        print(f'Pipeline SUCCEEDED')
    else:
        print(f'Pipeline {status.state}: {result.get("error", "unknown error")}')
    
    print(f'Would notify: {notification_email}')
    print(f'Full status: {json.dumps(result, indent=2)}')
    
    return json.dumps(result)
````

### Pipeline Definition

Wire the components together. The `ExitHandler` ensures the notification runs regardless of success or failure. After verification, the pipeline branches: if the deployment is healthy, old models are undeployed; if verification fails, traffic is rolled back to the previous model so the endpoint stays healthy.

````python
DEPLOYED_MODEL_NAME = 'bert-sst2-pipeline'

@dsl.pipeline(
    name=f'{SERIES}-{EXPERIMENT}',
    description='Automated model swap on PSC private endpoint',
    pipeline_root=f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/pipeline_root',
)
def psc_model_swap_pipeline(
    project: str,
    region: str,
    endpoint_resource_name: str,
    model_resource_name: str,
    deployed_model_display_name: str,
    machine_type: str = 'n1-standard-4',
    min_replicas: int = 1,
    max_replicas: int = 2,
    wait_seconds: int = 15,
    notification_email: str = '',
):
    # Exit handler: notify on completion (success or failure)
    notify = notify_status(
        notification_email=notification_email,
    ).set_display_name('Notify Status')
    
    with dsl.ExitHandler(exit_task=notify, name='Model Swap with Notification'):
        
        # Step 1: Verify endpoint
        verify = verify_endpoint(
            project=project,
            region=region,
            endpoint_resource_name=endpoint_resource_name,
        ).set_display_name('Verify PSC Endpoint')
        
        # Step 2: Deploy new model at 0% traffic
        deploy = deploy_model(
            project=project,
            region=region,
            endpoint_resource_name=endpoint_resource_name,
            model_resource_name=model_resource_name,
            deployed_model_display_name=deployed_model_display_name,
            machine_type=machine_type,
            min_replicas=min_replicas,
            max_replicas=max_replicas,
        ).set_display_name('Deploy New Model (0% Traffic)').after(verify)
        
        # Step 3: Shift 100% traffic to new model
        shift = shift_traffic(
            project=project,
            region=region,
            endpoint_resource_name=endpoint_resource_name,
            target_display_name=deploy.output,
            wait_seconds=wait_seconds,
        ).set_display_name('Shift Traffic to New Model').after(deploy)
        
        # Step 4: Verify deployment via management API
        check = verify_deployment(
            project=project,
            region=region,
            endpoint_resource_name=endpoint_resource_name,
            target_display_name=deploy.output,
        ).set_display_name('Verify Deployment').after(shift)
        
        # Step 5a: Undeploy old models (only if deployment verified)
        with dsl.If(check.output == True, name='Deployment Verified'):
            undeploy = undeploy_old_models(
                project=project,
                region=region,
                endpoint_resource_name=endpoint_resource_name,
                keep_display_name=deploy.output,
            ).set_display_name('Undeploy Old Models')
        
        # Step 5b: Roll back traffic if verification failed
        with dsl.Else(name='Verification Failed - Rollback'):
            rollback = rollback_traffic(
                project=project,
                region=region,
                endpoint_resource_name=endpoint_resource_name,
                failed_display_name=deploy.output,
                wait_seconds=wait_seconds,
            ).set_display_name('Rollback Traffic to Previous Model')
````
