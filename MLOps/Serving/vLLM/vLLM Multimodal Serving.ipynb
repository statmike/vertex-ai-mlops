# vLLM Multimodal Serving

Serve [Gemma 4](https://ai.google.dev/gemma/docs/core)'s native multimodal capabilities — image + text inference with vLLM. Gemma 4 models are natively multimodal, accepting images alongside text prompts without requiring separate vision encoders or preprocessing pipelines.

**What you'll learn:**
- Gemma 4 multimodal capabilities and which models support which modalities
- Image + text inference via the OpenAI vision API
- Use cases: image captioning, visual Q&A, document understanding, chart reading
- Structured output from images: extracting JSON from visual content
- Performance considerations: resolution, tokens per image, VRAM impact

**Model:** [google/gemma-4-E4B-it](https://huggingface.co/google/gemma-4-E4B-it) — 4B parameter dense model (~8 GB VRAM). Supports text, image, video, and audio. Chosen for its full multimodal support at a manageable size.

**Prerequisites:** A GPU with at least 8 GB VRAM (E4B is ~8 GB in bfloat16). GPU options include [Colab](https://colab.research.google.com/) (free T4), [Colab Enterprise](https://cloud.google.com/colab/docs/introduction) (L4/A100), [Vertex AI Workbench](https://cloud.google.com/vertex-ai/docs/workbench/introduction) (configurable GPU), [Cloud Workstations](https://cloud.google.com/workstations/docs/overview) (IDE of choice with attached GPU), or any local machine with a supported NVIDIA GPU.

## Gemma 4 Multimodal Capabilities

All [Gemma 4 models](https://huggingface.co/collections/google/gemma-4-release-680f5e77305ea7f903056ff8) are natively multimodal — they accept mixed content types (text, images, video, audio) in a single prompt. However, modality support varies by model size:

| Model | Text | Image | Video | Audio |
|-------|:---:|:---:|:---:|:---:|
| gemma-4-E2B-it (2B) | Yes | Yes | Yes | Yes |
| gemma-4-E4B-it (4B) | Yes | Yes | Yes | Yes |
| gemma-4-26B-A4B-it (26B MoE) | Yes | Yes | No | No |
| gemma-4-31B-it (31B) | Yes | Yes | No | No |

The smaller E2B and E4B dense models support all four modalities. The larger 26B MoE and 31B dense models support text + image only. We use **E4B** in this notebook for its full multimodal support at a manageable size (~8 GB VRAM at BF16).

vLLM serves multimodal models through the same OpenAI-compatible API — the only difference is that message content includes `image_url` entries alongside `text`. No special server configuration is needed beyond `--limit-mm-per-prompt` to control how many images a single request can contain.

## Create Test Images

Generate simple test images using PIL so the notebook is fully self-contained — no external downloads needed. These images are intentionally simple (text on colored backgrounds) to demonstrate the multimodal API without requiring large image files.

```python
def create_test_image(text, color='white', size=(400, 300)):
    """Create a simple test image with text."""
    img = Image.new('RGB', size, color)
    draw = ImageDraw.Draw(img)
    # Draw centered text
    bbox = draw.textbbox((0, 0), text)
    x = (size[0] - bbox[2]) // 2
    y = (size[1] - bbox[3]) // 2
    draw.text((x, y), text, fill='black')
    return img


def image_to_base64(img):
    """Convert PIL Image to base64 string."""
    buffer = io.BytesIO()
    img.save(buffer, format='PNG')
    return base64.b64encode(buffer.getvalue()).decode()


# Create test images
chart_img = create_test_image(
    'Sales Chart\n2024: $1.2M\n2025: $1.8M\n2026: $2.5M',
    color='lightyellow', size=(400, 300),
)
receipt_img = create_test_image(
    'RECEIPT\nCoffee: $4.50\nSandwich: $8.99\nTax: $1.08\nTotal: $14.57',
    color='white', size=(300, 400),
)
scene_img = create_test_image(
    'Mountain\nLandscape\n[Scenic View]',
    color='lightblue', size=(500, 300),
)

# Save to disk for reference
chart_img.save(f'{DIR}/chart.png')
receipt_img.save(f'{DIR}/receipt.png')
scene_img.save(f'{DIR}/scene.png')
print('Test images created and saved.')
```

```python
# You can also use images from URLs — the OpenAI vision API supports both formats:
#
# Base64 (used in this notebook — self-contained, no network needed):
#   {'type': 'image_url', 'image_url': {'url': f'data:image/png;base64,{img_b64}'}}
#
# URL (for images hosted online):
#   {'type': 'image_url', 'image_url': {'url': 'https://example.com/image.png'}}

print('Both base64 and URL image formats are supported.')
```

## Image + Text Inference

Send an image alongside a text prompt using the OpenAI vision API format. The `content` field in the user message is a list of parts — each part is either `text` or `image_url`. vLLM handles the multimodal encoding internally.

```python
client = OpenAI(base_url=f'http://localhost:{VLLM_PORT}/v1', api_key='unused')

# Basic image + text query
img_b64 = image_to_base64(scene_img)

response = client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {
            'role': 'user',
            'content': [
                {'type': 'image_url', 'image_url': {'url': f'data:image/png;base64,{img_b64}'}},
                {'type': 'text', 'text': 'Describe this image in detail.'},
            ],
        }
    ],
    max_tokens=256,
)
print(response.choices[0].message.content)
```

## Structured Output from Images

Combine vision capabilities with structured output to extract machine-readable data from images. Use `response_format={'type': 'json_object'}` to constrain the model to valid JSON output. For more on structured output (JSON mode, JSON schema, guided decoding), see the [Fundamentals notebook](./vLLM%20-%20Fundamentals.ipynb).

```python
class ReceiptItem(BaseModel):
    name: str
    price: float

class Receipt(BaseModel):
    items: list[ReceiptItem]
    subtotal: float
    tax: float
    total: float

print('Receipt schema:')
print(json.dumps(Receipt.model_json_schema(), indent=2))
```

```python
response = client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {
            'role': 'system',
            'content': 'Extract receipt data as JSON with fields: items (list of {name, price}), subtotal, tax, total.',
        },
        {
            'role': 'user',
            'content': [
                {'type': 'image_url', 'image_url': {'url': f'data:image/png;base64,{image_to_base64(receipt_img)}'}},
                {'type': 'text', 'text': 'Extract all items and totals from this receipt as JSON.'},
            ],
        }
    ],
    max_tokens=300,
    response_format={'type': 'json_object'},
)

result = json.loads(response.choices[0].message.content)
print(json.dumps(result, indent=2))
```

```python
# Validate against the Pydantic schema
try:
    receipt = Receipt(**result)
    print(f'Validated receipt with {len(receipt.items)} items:')
    for item in receipt.items:
        print(f'  {item.name}: ${item.price:.2f}')
    print(f'  Subtotal: ${receipt.subtotal:.2f}')
    print(f'  Tax: ${receipt.tax:.2f}')
    print(f'  Total: ${receipt.total:.2f}')
except Exception as e:
    print(f'Validation error: {e}')
    print('Raw output:', result)
```

## Batch Image Processing

Process multiple images efficiently by iterating through a list of image + question pairs. Each request is independent — for concurrent batch processing with `ThreadPoolExecutor`, see the throughput benchmarking section of the [Fundamentals notebook](./vLLM%20-%20Fundamentals.ipynb).

```python
images = [
    (scene_img, 'What is in this image?'),
    (chart_img, 'Summarize the data shown.'),
    (receipt_img, 'What is the total amount?'),
]

for img, question in images:
    response = client.chat.completions.create(
        model=MODEL_ID,
        messages=[
            {
                'role': 'user',
                'content': [
                    {'type': 'image_url', 'image_url': {'url': f'data:image/png;base64,{image_to_base64(img)}'}},
                    {'type': 'text', 'text': question},
                ],
            }
        ],
        max_tokens=100,
    )
    print(f'Q: {question}')
    print(f'A: {response.choices[0].message.content}\n')
```

## Multi-Turn Vision Conversations

Use multi-turn conversations to ask follow-up questions about an image. The image is sent in the first message, and subsequent messages can reference it.

```python
messages = [
    {
        'role': 'user',
        'content': [
            {'type': 'image_url', 'image_url': {'url': f'data:image/png;base64,{image_to_base64(chart_img)}'}},
            {'type': 'text', 'text': 'What does this chart show?'},
        ],
    }
]

# First turn
response = client.chat.completions.create(
    model=MODEL_ID, messages=messages, max_tokens=150,
)
answer_1 = response.choices[0].message.content
print(f'Turn 1: {answer_1}\n')

# Add assistant response and follow-up question
messages.append({'role': 'assistant', 'content': answer_1})
messages.append({'role': 'user', 'content': 'Which year had the highest growth rate?'})

# Second turn
response = client.chat.completions.create(
    model=MODEL_ID, messages=messages, max_tokens=150,
)
answer_2 = response.choices[0].message.content
print(f'Turn 2: {answer_2}')
```

## Performance Considerations

Multimodal inference introduces performance trade-offs that don't exist with text-only models:

| Factor | Impact | Recommendation |
|--------|--------|----------------|
| **Image resolution** | Higher resolution = more tokens = more VRAM and compute | Resize images to the minimum resolution needed for your task |
| **Images per request** | Each image adds to context length and KV cache | Use `--limit-mm-per-prompt image=N` to cap per-request images |
| **Token count per image** | Varies by resolution and model — can be 256-2048+ tokens per image | Monitor `prompt_tokens` in API responses to track actual usage |
| **VRAM overhead** | Image tokens consume KV cache space, reducing concurrent request capacity | Lower `--gpu-memory-utilization` if OOM errors occur with large images |
| **Batch throughput** | Mixed text+image batches are less uniform than text-only | For high throughput, keep image sizes consistent across batch requests |

**Tips for production:**
- Pre-process images to a consistent resolution before sending to the API
- Use lower `max_tokens` for tasks like captioning (100-200) vs. detailed analysis (500+)
- For document understanding at scale, consider splitting multi-page documents into individual page images
- Monitor `vllm:gpu_cache_usage_perc` — image-heavy workloads consume KV cache faster than text-only

```python
# Check token usage for different image sizes
sizes = [(200, 150), (400, 300), (800, 600)]

print(f'{"Image Size":<15} {"Prompt Tokens":>15} {"Completion Tokens":>18}')
print(f'{"-"*15} {"-"*15} {"-"*18}')

for size in sizes:
    test_img = create_test_image('Test', color='white', size=size)
    response = client.chat.completions.create(
        model=MODEL_ID,
        messages=[
            {
                'role': 'user',
                'content': [
                    {'type': 'image_url', 'image_url': {'url': f'data:image/png;base64,{image_to_base64(test_img)}'}},
                    {'type': 'text', 'text': 'What is this?'},
                ],
            }
        ],
        max_tokens=20,
    )
    usage = response.usage
    print(f'{size[0]}x{size[1]:<10} {usage.prompt_tokens:>15} {usage.completion_tokens:>18}')
```
