## What is vLLM?

Serving large language models efficiently is harder than serving traditional ML models. An LLM generates tokens one at a time (autoregressive decoding), maintaining a **KV cache** that grows with sequence length. Naive approaches waste GPU memory and leave the GPU idle between requests.

**vLLM** solves this with two key innovations:

### PagedAttention

Traditional serving pre-allocates a contiguous block of GPU memory for each request's KV cache, sized for the maximum possible sequence length. This wastes memory — most sequences are much shorter than the max. vLLM borrows the concept of **virtual memory paging** from operating systems:

- The KV cache is divided into fixed-size **blocks** (pages)
- Blocks are allocated on demand as the sequence grows
- Non-contiguous physical blocks are mapped via a block table (like a page table)
- When a sequence finishes, its blocks are immediately freed for reuse

This reduces KV cache memory waste from ~60-80% to near zero, allowing many more concurrent sequences.

### Continuous Batching

Traditional **static batching** waits until a batch of requests arrives, processes them together, and waits for the longest sequence to finish before accepting new requests. **Continuous batching** (also called iteration-level scheduling) is more dynamic:

- New requests can join the batch at any iteration (token generation step)
- Completed requests leave the batch immediately, freeing their KV cache blocks
- The GPU stays busy — no waiting for the slowest request in a batch

### OpenAI-Compatible API

vLLM exposes an API that matches the [OpenAI Chat Completions API](https://platform.openai.com/docs/api-reference/chat). This means you can use the standard `openai` Python client (or any OpenAI-compatible client) — just point `base_url` at your vLLM server. No new SDK to learn.

### How vLLM Compares to Other Approaches

| Approach | What it does | When to use it |
|----------|-------------|----------------|
| **FastAPI + transformers** | Wrap HuggingFace `pipeline()` in a REST endpoint | Prototyping, small models |
| **vLLM** | Purpose-built LLM serving with PagedAttention | LLM serving (default choice) |
| **Triton Inference Server** | General-purpose model serving (any framework) | Mixed workloads, ensemble pipelines |
| **TGI (Text Generation Inference)** | HuggingFace's LLM serving solution | Alternative to vLLM |

**Triton** is a general-purpose inference server — it can serve any ML model (TensorFlow, PyTorch, ONNX, TensorRT). For LLMs, Triton uses backends like TensorRT-LLM or the vLLM backend. **vLLM** is purpose-built for LLMs, making it simpler to set up and a better default choice for pure LLM serving.

## Start vLLM Server

We launch vLLM as a background subprocess using its OpenAI-compatible API server entrypoint. Key flags:

| Flag | Purpose |
|------|---------|
| `--model` | HuggingFace model ID or local path |
| `--port` | HTTP port for the API server |
| `--max-model-len` | Maximum sequence length (prompt + completion). Lower values save memory. |
| `--gpu-memory-utilization` | Fraction of GPU memory vLLM can use (0.0-1.0). The rest is left for PyTorch and system overhead. |
| `--dtype` | Weight precision. `auto` picks the model's native dtype (usually bfloat16). |
| `--trust-remote-code` | Allow running custom model code from HuggingFace. Required by some models. |

The server downloads the model on first run (cached in `~/.cache/huggingface/`), loads weights onto the GPU, pre-allocates KV cache blocks, and starts listening for HTTP requests.

````python
# Verify GPU is available before starting vLLM
import torch

if not torch.cuda.is_available():
    raise RuntimeError(
        'No GPU detected. vLLM requires a CUDA-capable GPU.\n'
        'GPU options:\n'
        '  - Google Colab: Runtime > Change runtime type > T4 GPU (free)\n'
        '  - Colab Enterprise: L4 or A100 runtimes\n'
        '  - Vertex AI Workbench: Create instance with GPU\n'
        '  - Cloud Workstations: Attach GPU to workstation config\n'
        '  - Local: NVIDIA GPU with CUDA drivers installed'
    )
print(f'GPU: {torch.cuda.get_device_name(0)} ({torch.cuda.get_device_properties(0).total_mem / 1e9:.1f} GB)')
````

````python
# Start vLLM server as a background process
vllm_process = subprocess.Popen(
    [
        sys.executable, '-m', 'vllm.entrypoints.openai.api_server',
        '--model', MODEL_ID,
        '--port', str(VLLM_PORT),
        '--max-model-len', '4096',
        '--gpu-memory-utilization', '0.85',
        '--dtype', 'auto',
        '--trust-remote-code',
    ],
    stdout=open(f'{DIR}/vllm_server.log', 'w'),
    stderr=subprocess.STDOUT,
)
print(f'vLLM server starting (PID: {vllm_process.pid}) ...')
````

````python
# Wait for the server to be ready
print('Waiting for vLLM server to be ready ...')
for i in range(120):
    try:
        resp = httpx.get(f'http://localhost:{VLLM_PORT}/health', timeout=2)
        if resp.status_code == 200:
            print(f'vLLM server is ready! (took ~{i * 5}s)')
            break
    except Exception:
        pass
    time.sleep(5)
else:
    print('WARNING: Server did not become ready in time. Check logs:')
    print(open(f'{DIR}/vllm_server.log').read()[-2000:])
````

## OpenAI-Compatible API

Once the server is running, we use the standard `openai` Python client. The only difference from calling OpenAI's API is the `base_url` — it points to our local vLLM server instead of `api.openai.com`. The `api_key` is required by the client library but ignored by vLLM (set to any string).

````python
client = OpenAI(base_url=VLLM_BASE_URL, api_key='unused')

# List available models
models = client.models.list()
print('Available models:', [m.id for m in models.data])
````

````python
# Simple chat completion
response = client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {'role': 'user', 'content': 'Explain PagedAttention in one paragraph.'}
    ],
    max_tokens=256,
    temperature=0.7,
)

print(response.choices[0].message.content)
print(f'\nTokens: {response.usage.prompt_tokens} prompt + {response.usage.completion_tokens} completion')
````

The response object matches the [OpenAI ChatCompletion schema](https://platform.openai.com/docs/api-reference/chat/object) exactly:

- `choices[0].message.content` — the generated text
- `choices[0].finish_reason` — why generation stopped (`stop`, `length`, `tool_calls`)
- `usage.prompt_tokens` / `usage.completion_tokens` — token counts for monitoring
- `model` — the model ID that served the request

This compatibility means any code written for OpenAI can switch to vLLM by changing only the `base_url`.

## Streaming

Streaming delivers tokens as they are generated, rather than waiting for the entire response. This is essential for interactive applications where **time-to-first-token (TTFT)** matters — the user sees output immediately instead of staring at a blank screen.

With `stream=True`, the API returns an iterator of `ChatCompletionChunk` objects. Each chunk contains a `delta` with the incremental content.

````python
# Streaming chat completion — measure time-to-first-token
start = time.time()
first_token_time = None

stream = client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {'role': 'user', 'content': 'Write a haiku about machine learning inference.'}
    ],
    max_tokens=100,
    stream=True,
)

tokens = []
for chunk in stream:
    if chunk.choices[0].delta.content:
        if first_token_time is None:
            first_token_time = time.time()
        tokens.append(chunk.choices[0].delta.content)
        print(chunk.choices[0].delta.content, end='', flush=True)

elapsed = time.time() - start
ttft = first_token_time - start if first_token_time else None
print(f'\n\nTime-to-first-token: {ttft:.3f}s')
print(f'Total time: {elapsed:.3f}s')
print(f'Tokens generated: {len(tokens)}')
print(f'Tokens/sec: {len(tokens)/elapsed:.1f}')
````

**Key metrics:**

| Metric | What it measures | Why it matters |
|--------|-----------------|----------------|
| **TTFT** (Time to First Token) | Latency from request to first generated token | User-perceived responsiveness |
| **Tokens/sec** | Generation throughput | Overall speed |
| **Total time** | End-to-end request time | Capacity planning |

TTFT is dominated by the **prefill** phase — processing the entire prompt through the model. After that, each token is generated in the **decode** phase, which is much faster per token.

## Sampling Parameters

Sampling parameters control how the model selects the next token from its probability distribution. These are the most important knobs for controlling output quality and creativity.

| Parameter | Default | Effect |
|-----------|---------|--------|
| `temperature` | 1.0 | Scales logits before softmax. 0 = greedy (deterministic), >1 = more random |
| `top_p` | 1.0 | Nucleus sampling — only consider tokens whose cumulative probability exceeds `top_p` |
| `top_k` | -1 (off) | Only consider the top K most likely tokens |
| `max_tokens` | model-dependent | Maximum number of tokens to generate |
| `stop` | None | Stop sequences — generation halts when any of these strings is produced |
| `frequency_penalty` | 0.0 | Penalize tokens proportional to how often they've appeared (reduces repetition) |
| `presence_penalty` | 0.0 | Penalize tokens that have appeared at all (encourages topic diversity) |

````python
# Compare different temperatures
for temp in [0.0, 0.7, 1.2]:
    response = client.chat.completions.create(
        model=MODEL_ID,
        messages=[{'role': 'user', 'content': 'Complete this sentence: The future of AI is'}],
        max_tokens=50,
        temperature=temp,
    )
    print(f'Temperature {temp}: {response.choices[0].message.content}\n')
````

````python
# Stop sequences — halt generation at specific strings
response = client.chat.completions.create(
    model=MODEL_ID,
    messages=[{'role': 'user', 'content': 'Count from 1 to 20, separated by commas.'}],
    max_tokens=200,
    temperature=0.0,
    stop=[', 11'],  # Stop before reaching 11
)
print(f'Stopped at: {response.choices[0].message.content}')
print(f'Finish reason: {response.choices[0].finish_reason}')
````

````python
# Frequency penalty — reduce repetition
for penalty in [0.0, 1.0, 2.0]:
    response = client.chat.completions.create(
        model=MODEL_ID,
        messages=[{'role': 'user', 'content': 'Write a short paragraph about cats.'}],
        max_tokens=80,
        temperature=0.7,
        frequency_penalty=penalty,
    )
    print(f'frequency_penalty={penalty}: {response.choices[0].message.content}\n')
````

**Practical guidance:**

- **Factual Q&A / extraction:** `temperature=0` (greedy, deterministic)
- **Creative writing / brainstorming:** `temperature=0.7-1.0`, `top_p=0.9`
- **Code generation:** `temperature=0-0.2` (precision matters)
- **Reducing repetition:** Start with `frequency_penalty=0.3-0.5`

## Concurrent Requests & Continuous Batching

One of vLLM's biggest advantages is its ability to handle many concurrent requests efficiently. Thanks to **continuous batching**, vLLM dynamically adds new requests to the running batch at each decode step. This means:

1. A new request arrives while other requests are mid-generation
2. vLLM runs the new request's prefill and adds it to the batch
3. All requests in the batch generate their next token together
4. When any request finishes, its KV cache blocks are freed immediately

The result: the GPU stays busy, and total throughput scales much better than sequential processing.

Let's measure this by sending multiple requests concurrently vs. sequentially.

````python
prompts = [
    'What is the capital of France?',
    'Explain quantum computing in one sentence.',
    'Write a Python function to reverse a string.',
    'What causes rain?',
    'Name three prime numbers greater than 100.',
]

def send_request(prompt):
    start = time.time()
    response = client.chat.completions.create(
        model=MODEL_ID,
        messages=[{'role': 'user', 'content': prompt}],
        max_tokens=100,
    )
    elapsed = time.time() - start
    return prompt[:40], response.choices[0].message.content[:80], elapsed

# Send all requests concurrently
start_all = time.time()
with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
    results = list(executor.map(send_request, prompts))
total = time.time() - start_all

print(f'Sent {len(prompts)} requests concurrently, completed in {total:.2f}s total\n')
for prompt_preview, response_preview, elapsed in results:
    print(f'  [{elapsed:.2f}s] {prompt_preview}...')
    print(f'         -> {response_preview}...\n')
````

````python
# Compare with sequential execution
start_seq = time.time()
for prompt in prompts:
    send_request(prompt)
sequential = time.time() - start_seq

print(f'Sequential: {sequential:.2f}s | Concurrent: {total:.2f}s | Speedup: {sequential/total:.1f}x')
````

The concurrent total is significantly less than the sum of sequential times because vLLM processes tokens for all requests in parallel. The speedup depends on:

- **GPU compute capacity** — larger GPUs can handle bigger batches
- **KV cache memory** — more memory = more concurrent sequences
- **Sequence lengths** — shorter sequences finish faster, freeing resources

In production, this means a single vLLM server can handle many concurrent users with much better throughput than processing requests one at a time.

## LoRA Adapters

**LoRA (Low-Rank Adaptation)** is a parameter-efficient fine-tuning method. Instead of updating all model weights during fine-tuning, LoRA freezes the base weights and trains small **adapter matrices** (typically <1% of total parameters). At inference time:

- The base model weights stay in GPU memory (shared across all adapters)
- Each LoRA adapter adds a small set of additional weights
- vLLM routes each request to the correct adapter based on the `model` field

This means you can serve **multiple fine-tuned variants** from a single base model, saving GPU memory compared to loading each fine-tuned model separately.

### vLLM LoRA Support

To enable LoRA in vLLM, add `--enable-lora` and specify adapters with `--lora-modules`:

```bash
vllm serve google/gemma-4-E2B-it \
    --enable-lora \
    --lora-modules \
        customer-support=/path/to/customer-support-adapter \
        code-review=/path/to/code-review-adapter
```

Then use the adapter name as the `model` parameter in requests:

```python
# Use a specific LoRA adapter
response = client.chat.completions.create(
    model='customer-support',  # routes to the customer-support adapter
    messages=[{'role': 'user', 'content': 'How do I reset my password?'}],
)

# Use the base model (no adapter)
response = client.chat.completions.create(
    model='google/gemma-4-E2B-it',  # routes to the base model
    messages=[{'role': 'user', 'content': 'What is quantum computing?'}],
)
```

## Structured Output

For many applications, you need the model to return structured data (JSON) rather than free-form text. vLLM supports two approaches:

1. **JSON mode** (`response_format={'type': 'json_object'}`) — instructs the model to produce valid JSON. The model tries its best, but there's no guarantee it matches a specific schema.

2. **Guided decoding** (`extra_body={'guided_json': ...}`) — constrains the token sampling at each step to guarantee the output matches a JSON schema. This uses grammar-based decoding under the hood (e.g., outlines or lm-format-enforcer).

### JSON Mode

````python
class SentimentResult(BaseModel):
    text: str
    sentiment: str  # positive, negative, or neutral
    confidence: float
    reasoning: str
````

````python
# JSON mode — model returns valid JSON
response = client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {
            'role': 'system',
            'content': (
                'You analyze sentiment. Always respond with valid JSON containing: '
                'text, sentiment (positive/negative/neutral), confidence (0-1), and reasoning.'
            ),
        },
        {
            'role': 'user',
            'content': 'Analyze: "The new restaurant downtown is absolutely amazing, best meal I\'ve had in years!"',
        },
    ],
    max_tokens=200,
    response_format={'type': 'json_object'},
)

result = json.loads(response.choices[0].message.content)
print(json.dumps(result, indent=2))

# Validate with Pydantic
validated = SentimentResult(**result)
print(f'\nValidated: sentiment={validated.sentiment}, confidence={validated.confidence}')
````

### Guided Decoding (Schema Enforcement)

Guided decoding goes further — it **constrains** the model's token generation to guarantee the output matches a JSON schema. At each decode step, vLLM masks out tokens that would violate the schema, so the output is always valid.

This is passed via `extra_body` since it's a vLLM-specific extension beyond the OpenAI API.

````python
# JSON Schema enforcement — guaranteed structure via guided decoding
person_schema = {
    'type': 'object',
    'properties': {
        'name': {'type': 'string'},
        'age': {'type': 'integer'},
        'company': {'type': 'string'},
        'role': {'type': 'string'},
        'location': {'type': 'string'},
    },
    'required': ['name', 'age', 'company', 'role', 'location'],
}

response = client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {
            'role': 'user',
            'content': (
                'Extract information from: "John Smith, age 35, works at Google '
                'as a software engineer in Mountain View, CA"'
            ),
        }
    ],
    max_tokens=200,
    extra_body={
        'guided_json': json.dumps(person_schema),
    },
)

extracted = json.loads(response.choices[0].message.content)
print(json.dumps(extracted, indent=2))
````

### JSON Mode vs. Guided Decoding

| Aspect | JSON Mode | Guided Decoding |
|--------|-----------|----------------|
| **Guarantee** | Valid JSON (best-effort structure) | Exact schema match (guaranteed) |
| **How it works** | Prompt instruction + constrained start/end tokens | Token-level masking at each decode step |
| **Performance** | No overhead | Slight overhead from grammar checking |
| **Schema control** | Model decides the keys | You define the exact schema |
| **API** | Standard OpenAI `response_format` | vLLM extension `extra_body.guided_json` |
| **Use when** | Flexible JSON output is fine | You need guaranteed structure (APIs, databases) |

## Tool Calling

Tool calling (also called function calling) lets the model request that external functions be executed. The flow:

1. **Define tools** — describe available functions (name, description, parameters) in the request
2. **Model decides** — based on the user's message, the model either responds directly or emits one or more `tool_calls`
3. **Execute tools** — your application runs the requested functions and collects results
4. **Feed results back** — send the tool results to the model for a final response

The model never executes the functions itself — it only decides which function to call and with what arguments. Your code handles execution.

````python
# Define tools (functions the model can call)
tools = [
    {
        'type': 'function',
        'function': {
            'name': 'get_weather',
            'description': 'Get the current weather for a location',
            'parameters': {
                'type': 'object',
                'properties': {
                    'location': {
                        'type': 'string',
                        'description': 'City and state, e.g., San Francisco, CA',
                    },
                    'unit': {
                        'type': 'string',
                        'enum': ['celsius', 'fahrenheit'],
                        'description': 'Temperature unit',
                    },
                },
                'required': ['location'],
            },
        },
    },
    {
        'type': 'function',
        'function': {
            'name': 'search_web',
            'description': 'Search the web for information',
            'parameters': {
                'type': 'object',
                'properties': {
                    'query': {
                        'type': 'string',
                        'description': 'The search query',
                    },
                },
                'required': ['query'],
            },
        },
    },
]

# Ask something that should trigger a tool call
response = client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {'role': 'user', 'content': "What's the weather like in San Francisco today?"}
    ],
    tools=tools,
    tool_choice='auto',
    max_tokens=200,
)

message = response.choices[0].message
if message.tool_calls:
    for tc in message.tool_calls:
        print(f'Tool call: {tc.function.name}({tc.function.arguments})')
else:
    print(f'Response: {message.content}')
````

### Complete Tool Calling Flow

In a real application, you would execute the tool and feed the result back to the model:

````python
# Simulate a complete tool calling flow
def get_weather(location, unit='fahrenheit'):
    """Simulated weather function."""
    return {'location': location, 'temperature': 68, 'unit': unit, 'condition': 'sunny'}

def search_web(query):
    """Simulated web search."""
    return {'results': [f'Result about: {query}']}

tool_functions = {'get_weather': get_weather, 'search_web': search_web}

# Step 1: Send user message with tools
messages = [
    {'role': 'user', 'content': "What's the weather in San Francisco?"}
]

response = client.chat.completions.create(
    model=MODEL_ID,
    messages=messages,
    tools=tools,
    tool_choice='auto',
    max_tokens=200,
)

assistant_message = response.choices[0].message

# Step 2: If the model wants to call tools, execute them
if assistant_message.tool_calls:
    # Add the assistant's response (with tool calls) to the conversation
    messages.append(assistant_message)

    # Execute each tool call and add the results
    for tc in assistant_message.tool_calls:
        func = tool_functions[tc.function.name]
        args = json.loads(tc.function.arguments)
        result = func(**args)
        print(f'Executed: {tc.function.name}({args}) -> {result}')

        messages.append({
            'role': 'tool',
            'tool_call_id': tc.id,
            'content': json.dumps(result),
        })

    # Step 3: Get the final response with tool results
    final_response = client.chat.completions.create(
        model=MODEL_ID,
        messages=messages,
        tools=tools,
        max_tokens=200,
    )
    print(f'\nFinal response: {final_response.choices[0].message.content}')
else:
    print(f'Direct response (no tool call): {assistant_message.content}')
````

### Tool Calling Notes

- **Model support varies** — not all models support tool calling well. Gemma 4 instruction-tuned models include tool calling capabilities.
- **`tool_choice`** controls behavior: `'auto'` (model decides), `'none'` (never call tools), or `{'type': 'function', 'function': {'name': '...'}}` (force a specific tool).
- **Parallel tool calls** — the model can request multiple tool calls in a single response.
- **Validation** — always validate the model's function arguments before execution.

## Speculative Decoding

**Speculative decoding** is a latency optimization technique. The idea:

1. A small, fast **draft model** proposes N candidate tokens autoregressively
2. The large **target model** verifies all N tokens in a **single forward pass** (parallel verification)
3. Accepted tokens skip the expensive autoregressive generation; rejected tokens are re-generated by the target model

Since verifying N tokens in one forward pass is much cheaper than generating N tokens autoregressively, this can significantly reduce latency when the draft model's predictions are accepted (high acceptance rate).

### When Does Speculative Decoding Help?

| Scenario | Benefit |
|----------|--------|
| Large target model, small draft model available | High — draft model runs fast, target model verifies cheaply |
| Code generation, structured output | High — predictable tokens, high acceptance rate |
| Creative writing, open-ended generation | Lower — less predictable tokens, lower acceptance rate |
| Already memory-constrained | Negative — draft model uses additional GPU memory |

### vLLM Speculative Decoding Configuration

Speculative decoding requires server-level configuration (it cannot be enabled per-request). You restart vLLM with additional flags:

````python
# Speculative decoding server configuration (reference code)
#
# This requires restarting the server, so we show the command pattern here.
#
# Example 1: Use a smaller model from the same family as the draft
# vllm serve google/gemma-4-26B-A4B-it \
#     --speculative-model google/gemma-4-E2B-it \
#     --num-speculative-tokens 5 \
#     --port 8000
#
# Example 2: Using ngram-based speculation (no draft model needed)
# vllm serve google/gemma-4-E2B-it \
#     --speculative-model [ngram] \
#     --ngram-prompt-lookup-max 4 \
#     --num-speculative-tokens 5 \
#     --port 8000
#
# Key flags:
# --speculative-model       : HF model ID for the draft model, or [ngram] for ngram-based
# --num-speculative-tokens  : Number of tokens the draft proposes per step (typically 3-7)
# --ngram-prompt-lookup-max : For ngram speculation, max ngram length to match from prompt

print('Speculative decoding patterns shown above (commented).')
print('Requires server restart with --speculative-model and --num-speculative-tokens flags.')
print('\nOnce enabled, the API is identical — speculative decoding is transparent to clients.')
````

### How Speculative Decoding Works (Step by Step)

```
Prompt: "The capital of France is"

Step 1 - Draft model proposes 5 tokens:
  [Paris] [,] [which] [is] [known]

Step 2 - Target model verifies all 5 in one forward pass:
  [Paris] ✓  [,] ✓  [a] ✗  (rejected "which", target says "a")

Step 3 - Accept verified tokens, re-generate from rejection point:
  Output: "Paris, a" (3 tokens from one target model forward pass)
  
Without speculation: 3 forward passes for 3 tokens
With speculation: 1 forward pass for 3 tokens (+ 5 cheap draft passes)
```

The speedup depends on the **acceptance rate** — how often the draft model's predictions match what the target model would have generated. Higher acceptance rate = more tokens verified per forward pass = lower latency.

## Benchmarking

Let's measure the key performance characteristics of our vLLM deployment:

1. **Throughput** — total tokens/sec across concurrent requests
2. **Latency** — time per request
3. **TTFT** — time to first token (streaming)

````python
# Throughput benchmark
def benchmark(num_requests=10, max_tokens=100):
    """Send concurrent requests and measure throughput."""
    prompt = 'Write a brief explanation of how neural networks learn from data.'

    def single_request(_):
        start = time.time()
        response = client.chat.completions.create(
            model=MODEL_ID,
            messages=[{'role': 'user', 'content': prompt}],
            max_tokens=max_tokens,
        )
        elapsed = time.time() - start
        return response.usage.completion_tokens, elapsed

    start_all = time.time()
    with concurrent.futures.ThreadPoolExecutor(max_workers=num_requests) as executor:
        results = list(executor.map(single_request, range(num_requests)))
    total_time = time.time() - start_all

    total_tokens = sum(r[0] for r in results)
    avg_latency = sum(r[1] for r in results) / len(results)

    print(f'Requests:        {num_requests}')
    print(f'Total time:      {total_time:.2f}s')
    print(f'Total tokens:    {total_tokens}')
    print(f'Throughput:      {total_tokens / total_time:.1f} tokens/sec')
    print(f'Avg latency:     {avg_latency:.2f}s per request')
    print(f'Requests/sec:    {num_requests / total_time:.1f}')

benchmark(num_requests=5, max_tokens=100)
````

````python
# TTFT benchmark
def ttft_benchmark(num_requests=5):
    """Measure time-to-first-token across multiple requests."""
    prompt = 'Explain in detail how gradient descent works in neural network training.'
    ttfts = []

    for i in range(num_requests):
        start = time.time()
        stream = client.chat.completions.create(
            model=MODEL_ID,
            messages=[{'role': 'user', 'content': prompt}],
            max_tokens=50,
            stream=True,
        )
        for chunk in stream:
            if chunk.choices[0].delta.content:
                ttfts.append(time.time() - start)
                # Consume rest of stream
                for _ in stream:
                    pass
                break

    print(f'TTFT over {num_requests} requests:')
    print(f'  Min:    {min(ttfts):.3f}s')
    print(f'  Max:    {max(ttfts):.3f}s')
    print(f'  Mean:   {sum(ttfts) / len(ttfts):.3f}s')
    print(f'  Median: {sorted(ttfts)[len(ttfts) // 2]:.3f}s')

ttft_benchmark()
````

### Interpreting Benchmark Results

- **Throughput (tokens/sec)** — The total generation speed across all concurrent requests. Higher is better. This scales with GPU compute and batch size.
- **Avg latency** — The average time per individual request. Note this is higher with concurrent requests because the GPU is shared.
- **TTFT** — Time to first token. This is dominated by the prefill phase (processing the prompt). Longer prompts = higher TTFT.

For production deployment, you'll typically want to run vLLM's built-in benchmark tool for more rigorous testing:

## Additional vLLM Features

Beyond the features covered in detail above, vLLM supports several other capabilities worth knowing about:

### Multi-GPU / Tensor Parallelism

For models that don't fit on a single GPU, vLLM can shard the model across multiple GPUs:

```bash
vllm serve google/gemma-4-26B-A4B-it --tensor-parallel-size 4
```

This splits the model's weight matrices across GPUs and coordinates the computation. Each GPU holds a fraction of the weights and processes a slice of each layer.

### Quantization

vLLM supports several quantization methods to reduce memory usage and potentially improve throughput:

```bash
# Use a pre-quantized GPTQ model
vllm serve TheBloke/Llama-2-7B-Chat-GPTQ --quantization gptq

# Use AWQ quantization
vllm serve TheBloke/Llama-2-7B-Chat-AWQ --quantization awq

# Use bitsandbytes for on-the-fly quantization
vllm serve google/gemma-4-E2B-it --quantization bitsandbytes --load-format bitsandbytes
```

Quantization trades a small amount of output quality for significant memory savings, allowing you to run larger models on smaller GPUs.

### Prefix Caching

When many requests share the same system prompt or prefix, **automatic prefix caching** avoids redundant computation:

```bash
vllm serve google/gemma-4-E2B-it --enable-prefix-caching
```

This caches the KV states for common prefixes. Subsequent requests with the same prefix skip the prefill for the cached portion, reducing TTFT.
