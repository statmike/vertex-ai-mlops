# vLLM on Cloud Run

Deploy vLLM to [Cloud Run](https://cloud.google.com/run) with L4 GPU for serverless LLM serving that scales to zero when idle. Cloud Run GPU support (GA since June 2025) makes it the simplest serverless option for LLM inference.

**What you'll learn:**
- Cloud Run GPU deployment with `nvidia-l4` accelerator
- Cold start analysis: vLLM + model loading time
- OIDC authentication for secure access
- SSE streaming with native Cloud Run support
- Scale-to-zero and cost implications
- Concurrency tuning: Cloud Run vs vLLM settings

**Model:** [google/gemma-4-26B-A4B-it](https://huggingface.co/google/gemma-4-26B-A4B-it) — 26B MoE at INT4 quantization, fits single L4.

**Prerequisites:** A GCP project with Cloud Run GPU quota in a supported region.

> For vLLM fundamentals (engine internals, PagedAttention, quantization), see [vLLM Fundamentals](./vLLM%20Fundamentals.ipynb).

## Cloud Run GPU Constraints

Cloud Run GPU support has specific constraints that shape how we deploy vLLM:

| Constraint | Detail |
|-----------|--------|
| **GPU type** | Single NVIDIA L4 only (24 GB VRAM) |
| **GA regions** | `us-central1`, `europe-west1`, `asia-southeast1`, `asia-northeast1` (verify [current list](https://cloud.google.com/run/docs/configuring/services/gpu)) |
| **Max model size** | ~16 GB weights at INT4 leaves ~8 GB for KV cache |
| **Startup probe** | Up to 30 min timeout (critical for large model loading) |
| **Multi-GPU** | Not supported — single GPU per instance |
| **Resources** | Up to 8 vCPU, 32 GiB memory per GPU instance |

These constraints make the L4 a good fit for quantized models up to ~26B parameters. The `google/gemma-4-26B-A4B-it` model uses a Mixture-of-Experts architecture with only 4B active parameters, keeping inference fast while fitting comfortably within the L4's VRAM.

## Container Image

Build a custom container that runs vLLM's OpenAI-compatible API server. The Dockerfile extends the official `vllm/vllm-openai` image and configures it for the Gemma model with INT4 quantization via `bitsandbytes`.

Key settings:
- `--port 8080` — Cloud Run's expected port
- `--max-model-len 4096` — limits KV cache memory usage
- `--gpu-memory-utilization 0.90` — reserves 10% VRAM headroom
- `--quantization bitsandbytes` — INT4 quantization for L4 VRAM budget

````python
%%writefile {DIR}/Dockerfile
FROM vllm/vllm-openai:latest

ENV HF_TOKEN=""

ENTRYPOINT ["python", "-m", "vllm.entrypoints.openai.api_server", \
    "--model", "google/gemma-4-26B-A4B-it", \
    "--port", "8080", \
    "--max-model-len", "4096", \
    "--gpu-memory-utilization", "0.90", \
    "--quantization", "bitsandbytes", \
    "--load-format", "bitsandbytes", \
    "--trust-remote-code"]
````

## Deploy to Cloud Run

Deploy the vLLM container to Cloud Run with an L4 GPU. Key configuration choices:

- **Scale to zero** (`min_instance_count=0`): no cost when idle, but cold starts on first request
- **Single instance** (`max_instance_count=1`): sufficient for development; increase for production
- **Concurrency 8**: matches vLLM's default continuous batching capacity on a single L4
- **Startup probe**: 40 failures x 30s period = 20 min max startup time for model download + loading
- **HF_TOKEN**: passed as env var for gated model access

The delete-then-create pattern ensures a clean deployment (avoids stale revision caching).

````python
# Delete existing service if present (ensures fresh deployment)
try:
    run_client.delete_service(name=SERVICE_FULL_NAME).result()
    print(f'Deleted previous service: {SERVICE_NAME}')
    time.sleep(10)
except Exception:
    pass
````

````python
service = run_v2.Service(
    template=run_v2.RevisionTemplate(
        containers=[
            run_v2.Container(
                image=image_uri,
                resources=run_v2.ResourceRequirements(
                    limits={'cpu': '8', 'memory': '32Gi', 'nvidia.com/gpu': '1'},
                ),
                ports=[run_v2.ContainerPort(container_port=8080)],
                startup_probe=run_v2.Probe(
                    http_get=run_v2.HTTPGetAction(path='/health', port=8080),
                    initial_delay_seconds=30,
                    period_seconds=30,
                    timeout_seconds=5,
                    failure_threshold=40,  # 40 * 30s = 20 min max startup
                ),
                env=[
                    run_v2.EnvVar(name='HF_TOKEN', value=os.environ.get('HF_TOKEN', '')),
                ],
            ),
        ],
        scaling=run_v2.RevisionScaling(
            min_instance_count=0,  # Scale to zero!
            max_instance_count=1,
        ),
        max_instance_request_concurrency=8,
        timeout=run_v2.types.Duration(seconds=300),
        node_selector=run_v2.NodeSelector(
            accelerator='nvidia-l4',
        ),
    ),
)

print(f'Creating Cloud Run service: {SERVICE_NAME} ...')
operation = run_client.create_service(
    parent=f'projects/{PROJECT_ID}/locations/{REGION}',
    service=service,
    service_id=SERVICE_NAME,
)
result = operation.result(timeout=1800)
SERVICE_URL = result.uri
print(f'Service deployed: {SERVICE_URL}')
````

## Wait for Service Ready

vLLM needs time to download the model from Hugging Face and load it onto the GPU. With scale-to-zero, the first request after idle triggers a cold start that includes:

1. Container image pull (~10 GB)
2. vLLM process startup
3. Model download from Hugging Face (~13 GB for Gemma 4 26B)
4. Model loading and GPU memory allocation

This can take 5-15 minutes depending on network conditions.

````python
# Get OIDC token for authenticated requests
credentials, _ = google.auth.default()
auth_req = google.auth.transport.requests.Request()
credentials.refresh(auth_req)
id_tok = google_id_token.fetch_id_token(auth_req, SERVICE_URL)

print('Waiting for vLLM to be ready ...')
for i in range(60):
    try:
        resp = httpx.get(
            f'{SERVICE_URL}/health',
            headers={'Authorization': f'Bearer {id_tok}'},
            timeout=10,
        )
        if resp.status_code == 200:
            print(f'vLLM is ready! (took ~{i * 15}s)')
            break
    except Exception:
        pass
    if i % 2 == 0:
        print(f'  Not yet ready ... ({i * 15}s)')
        # Refresh token if it might be expiring
        if i > 0 and i % 8 == 0:
            credentials.refresh(auth_req)
            id_tok = google_id_token.fetch_id_token(auth_req, SERVICE_URL)
    time.sleep(15)
````

---
## OIDC Authentication

Cloud Run services require authentication by default. Unlike Vertex AI (which uses OAuth2 access tokens), Cloud Run uses **OIDC ID tokens**:

- The token's **audience** must match the service URL
- Use `google.oauth2.id_token.fetch_id_token()` to obtain an ID token
- Pass it as a `Bearer` token in the `Authorization` header
- Tokens expire after ~1 hour — refresh for long-running sessions

The OpenAI SDK accepts the ID token in the `api_key` field, which it sends as `Authorization: Bearer <token>` on every request.

> **Public access:** To allow unauthenticated access, add an `allUsers` IAM binding with `roles/run.invoker`. This is not recommended for production.

## Inference via OpenAI SDK

vLLM exposes an OpenAI-compatible API, so the standard `openai` Python SDK works directly. Point it at the Cloud Run service URL and use the OIDC token as the API key.

````python
# Refresh OIDC token
id_tok = google_id_token.fetch_id_token(auth_req, SERVICE_URL)

# Configure OpenAI SDK to point at Cloud Run
client = OpenAI(
    base_url=f'{SERVICE_URL}/v1',
    api_key=id_tok,  # OIDC token as API key
)
````

````python
response = client.chat.completions.create(
    model='google/gemma-4-26B-A4B-it',
    messages=[
        {'role': 'user', 'content': 'What are the advantages of serverless LLM serving?'}
    ],
    max_tokens=256,
    temperature=0.7,
)
print(response.choices[0].message.content)
````

## SSE Streaming

Cloud Run natively supports Server-Sent Events (SSE), which vLLM uses for streaming responses. Streaming reduces time-to-first-token (TTFT) perception by delivering tokens as they are generated rather than waiting for the full response.

````python
start = time.time()
first_token = None
token_count = 0

stream = client.chat.completions.create(
    model='google/gemma-4-26B-A4B-it',
    messages=[
        {'role': 'user', 'content': 'Explain how Cloud Run scales GPU workloads.'}
    ],
    max_tokens=200,
    stream=True,
)

for chunk in stream:
    if chunk.choices[0].delta.content:
        if first_token is None:
            first_token = time.time()
        token_count += 1
        print(chunk.choices[0].delta.content, end='', flush=True)

elapsed = time.time() - start
ttft = first_token - start if first_token else None
tps = token_count / (elapsed - ttft) if ttft and elapsed > ttft else 0

print(f'\n\nTTFT: {ttft:.3f}s | Total: {elapsed:.3f}s | Tokens: {token_count} | Throughput: {tps:.1f} tok/s')
````

## Cold Start Analysis

With `min_instance_count=0`, Cloud Run spins down the instance after an idle period (~15 minutes). The next request triggers a **cold start** with these components:

| Phase | Duration | Notes |
|-------|----------|-------|
| Container pull | 1-3 min | ~10 GB vLLM image |
| vLLM startup | 10-30s | Python process + engine init |
| Model download | 2-8 min | ~13 GB from Hugging Face (first time or uncached) |
| GPU loading | 1-2 min | Quantize + allocate KV cache |
| **Total** | **5-15 min** | Varies by network and model size |

**Mitigation strategies:**
- Set `min_instance_count=1` to keep one instance warm (pay for always-on GPU)
- Pre-bake the model into the container image (larger image, but no HF download at startup)
- Use a GCS FUSE mount for model weights (consistent download speed within GCP)

Check the revision startup time from the Cloud Run API:

````python
# Check service status and revision details
svc = run_client.get_service(name=SERVICE_FULL_NAME)
print(f'Service:  {svc.name.split("/")[-1]}')
print(f'URL:      {svc.uri}')
print(f'Created:  {svc.create_time}')
print(f'Updated:  {svc.update_time}')

# List revisions to see startup timing
revisions_client = run_v2.RevisionsClient()
for rev in revisions_client.list_revisions(parent=SERVICE_FULL_NAME):
    print(f'\nRevision: {rev.name.split("/")[-1]}')
    print(f'  Created:  {rev.create_time}')
    print(f'  Image:    {rev.containers[0].image}')
    for condition in rev.conditions:
        print(f'  {condition.type_}: {condition.state.name} ({condition.reason})')
````

## Concurrency Tuning

Two concurrency settings interact when running vLLM on Cloud Run:

### Cloud Run: `max_instance_request_concurrency`

Controls how many concurrent requests Cloud Run routes to each container instance. When all instances are at capacity, Cloud Run scales up new instances (subject to `max_instance_count`).

### vLLM: `--max-num-seqs` (default: 256)

Controls how many sequences vLLM can process simultaneously via continuous batching. This is an internal engine limit.

### How they interact

```
Cloud Run concurrency = 8
         │
         ▼
┌─── Cloud Run Instance ────┐
│  Max 8 concurrent requests │
│  ┌─── vLLM Engine ─────┐  │
│  │  max-num-seqs: 256  │  │
│  │  Continuous batching │  │
│  │  (handles up to 8)  │  │
│  └─────────────────────┘  │
└───────────────────────────┘
```

**Rule of thumb:** Set Cloud Run concurrency <= vLLM's `--max-num-seqs`. For GPU workloads, a lower Cloud Run concurrency (4-16) prevents overloading the GPU while still allowing vLLM's continuous batching to work efficiently.

````python
import concurrent.futures

def send_request(prompt):
    """Send a single chat completion and return latency."""
    # Each thread gets a fresh token
    tok = google_id_token.fetch_id_token(auth_req, SERVICE_URL)
    thread_client = OpenAI(base_url=f'{SERVICE_URL}/v1', api_key=tok)
    start = time.time()
    resp = thread_client.chat.completions.create(
        model='google/gemma-4-26B-A4B-it',
        messages=[{'role': 'user', 'content': prompt}],
        max_tokens=50,
        temperature=0.7,
    )
    elapsed = time.time() - start
    return elapsed, resp.usage.completion_tokens


prompts = [
    'What is Cloud Run?',
    'What is vLLM?',
    'Explain GPU memory.',
    'What is KV cache?',
    'Define serverless.',
    'What is OIDC?',
    'Explain autoscaling.',
    'What is INT4 quantization?',
]

print(f'Sending {len(prompts)} concurrent requests ...')
overall_start = time.time()

with concurrent.futures.ThreadPoolExecutor(max_workers=len(prompts)) as executor:
    futures = [executor.submit(send_request, p) for p in prompts]
    results = [f.result() for f in futures]

overall_elapsed = time.time() - overall_start
latencies = sorted([r[0] for r in results])
total_tokens = sum(r[1] for r in results)

print(f'\nAll {len(prompts)} completed in {overall_elapsed:.1f}s')
print(f'Individual latencies:')
print(f'  Median: {latencies[len(latencies)//2]:.2f}s')
print(f'  Min:    {latencies[0]:.2f}s')
print(f'  Max:    {latencies[-1]:.2f}s')
print(f'Aggregate throughput: {total_tokens / overall_elapsed:.1f} tok/s')
````

## Scale-to-Zero

With `min_instance_count=0`, Cloud Run automatically scales the GPU instance to zero after an idle period (~15 minutes). This is the defining advantage of Cloud Run for LLM serving:

**Benefits:**
- Pay only for active inference time (GPU-seconds, not GPU-hours)
- Ideal for development, testing, and bursty workloads
- No manual instance management

**Trade-offs:**
- First request after idle triggers 5-15 min cold start
- Not suitable for latency-sensitive production workloads
- Model must be re-downloaded and re-loaded on each cold start (unless baked into the image)

**Always-warm alternative:**
```python
scaling=run_v2.RevisionScaling(
    min_instance_count=1,  # Keep one instance warm
    max_instance_count=3,  # Scale up to 3 for bursts
)
```
This eliminates cold starts but incurs continuous GPU cost (~$0.70/hr for L4 on Cloud Run).

**Cost comparison:**

| Pattern | Monthly cost (est.) | Latency |
|---------|--------------------|---------|
| Scale-to-zero, 2 hr/day usage | ~$42 | 5-15 min cold start |
| Always-warm (min=1) | ~$504 | Consistent low latency |
| GKE Autopilot (1x L4) | ~$450 | Consistent low latency |
