## Google's Pre-built vLLM Container

Google maintains an **optimized vLLM fork** packaged as a pre-built container for Vertex AI. This container includes:

- **Parallel model downloading** — downloads model shards from Hugging Face in parallel, significantly faster than sequential downloading

Using this container means **no custom Dockerfile needed**. You specify the container URI, pass vLLM args, and deploy. Google handles container maintenance, security patches, and optimization updates.

## Upload Model to Vertex AI

Upload the model to the Vertex AI Model Registry. This doesn't download the model yet — it registers the container image and vLLM serving arguments. The model will be downloaded at deploy time when the container starts.

Key arguments:
- **`--model`** — Hugging Face model ID. The container downloads the model at startup.
- **`--max-model-len`** — Maximum sequence length (prompt + generation). Lower values save KV cache memory. 4096 is a reasonable default for chat.
- **`--gpu-memory-utilization`** — Fraction of GPU memory for vLLM (model weights + KV cache). Default 0.90 reserves 10% for PyTorch overhead.
- **`--quantization bitsandbytes`** + **`--load-format bitsandbytes`** — INT4 quantization via bitsandbytes. Reduces the 26B MoE model from ~52 GB (BF16) to ~16 GB, fitting on a single L4 (24 GB).
- **`--trust-remote-code`** — Required for some Hugging Face models that include custom Python code.

> **HF_TOKEN:** The Gemma 4 models require accepting a license agreement on Hugging Face. Set your `HF_TOKEN` environment variable before running this notebook, or replace the empty string below.

````python
model = aiplatform.Model.upload(
    display_name=MODEL_DISPLAY_NAME,
    serving_container_image_uri=VLLM_CONTAINER_URI,
    serving_container_args=[
        '--model', MODEL_ID,
        '--max-model-len', '4096',
        '--gpu-memory-utilization', '0.90',
        '--dtype', 'auto',
        '--quantization', 'bitsandbytes',
        '--load-format', 'bitsandbytes',
        '--trust-remote-code',
    ],
    serving_container_ports=[8080],
    serving_container_predict_route='/v1/chat/completions',
    serving_container_health_route='/health',
    serving_container_environment_variables={
        'HF_TOKEN': os.environ.get('HF_TOKEN', ''),
    },
)
model.wait()
print(f'Model uploaded: {model.resource_name}')
````

## Deploy to Endpoint

Create a **dedicated public endpoint** and deploy the model. Dedicated endpoints provide a unique DNS name and are required for `rawPredict` — Vertex AI's mechanism for passing request bodies through to the container unchanged.

### Machine Type and GPU

- **`g2-standard-12`** — 12 vCPUs, 48 GB RAM, 1x NVIDIA L4 GPU (24 GB VRAM)
- The L4 is Google Cloud's most cost-effective inference GPU. At INT4 quantization, the Gemma 4 26B-A4B MoE model uses ~16 GB of the 24 GB available, leaving room for KV cache.
- For larger models or higher concurrency, consider `a2-highgpu-1g` (1x A100 80 GB) or multi-GPU configurations.

### Deployment Time

Deployment typically takes 10-20 minutes:
1. Vertex AI provisions the VM and attaches the GPU (~3-5 min)
2. The container pulls the model from Hugging Face (~5-10 min for 26B params)
3. vLLM loads and quantizes the model into GPU memory (~2-5 min)
4. Health check passes and traffic starts routing

## rawPredict

Vertex AI's standard `predict()` method expects the `{"instances": [...]}` format and parses the request body before forwarding it to the container. vLLM uses the **OpenAI API format** (`{"messages": [...]}`) which is incompatible with this parsing.

**`rawPredict`** bypasses Vertex AI's request parsing entirely — the request body is forwarded to the container unchanged. This is required for vLLM because:
1. The OpenAI chat completions format doesn't match Vertex AI's expected schema
2. vLLM needs to receive the exact OpenAI-format payload to route it correctly
3. Streaming (SSE) requires raw HTTP pass-through

We use `httpx` to send requests directly to the dedicated endpoint DNS with a GCP auth token.

````python
# Get GCP credentials for authentication
credentials, _ = google.auth.default()
credentials.refresh(google.auth.transport.requests.Request())

headers = {
    'Authorization': f'Bearer {credentials.token}',
    'Content-Type': 'application/json',
}

payload = {
    'model': MODEL_ID,
    'messages': [
        {'role': 'user', 'content': 'Explain the Mixture of Experts architecture in one paragraph.'}
    ],
    'max_tokens': 256,
    'temperature': 0.7,
}

# rawPredict sends the request body directly to the container
url = f'https://{endpoint_dns}/v1/chat/completions'

response = httpx.post(url, headers=headers, json=payload, timeout=60)
result = response.json()

print(result['choices'][0]['message']['content'])
print(f"\nTokens: {result['usage']['prompt_tokens']} prompt + {result['usage']['completion_tokens']} completion")
````

## streamRawPredict

Streaming is essential for LLM serving — users see tokens as they're generated rather than waiting for the entire response. vLLM streams responses using **Server-Sent Events (SSE)**, the same format the OpenAI API uses.

Set `"stream": true` in the request payload. The response is a stream of `data: {...}` lines, each containing a chunk with a `delta` field. The final line is `data: [DONE]`.

We measure **Time to First Token (TTFT)** — the latency from request sent to first token received. This is the most important latency metric for interactive LLM applications.

````python
# Refresh credentials (tokens expire after ~1 hour)
credentials.refresh(google.auth.transport.requests.Request())
headers['Authorization'] = f'Bearer {credentials.token}'

stream_payload = {
    'model': MODEL_ID,
    'messages': [
        {'role': 'user', 'content': 'What are three key advantages of deploying LLMs on managed cloud endpoints?'}
    ],
    'max_tokens': 512,
    'temperature': 0.7,
    'stream': True,
}

start = time.time()
first_token_time = None
token_count = 0

with httpx.stream('POST', url, headers=headers, json=stream_payload, timeout=120) as response:
    for line in response.iter_lines():
        if line.startswith('data: ') and line != 'data: [DONE]':
            chunk = json.loads(line[6:])
            content = chunk['choices'][0].get('delta', {}).get('content', '')
            if content:
                if first_token_time is None:
                    first_token_time = time.time()
                token_count += 1
                print(content, end='', flush=True)

elapsed = time.time() - start
ttft = first_token_time - start if first_token_time else None
tps = token_count / (elapsed - ttft) if ttft and elapsed > ttft else 0

print(f'\n\n--- Streaming Metrics ---')
print(f'TTFT: {ttft:.3f}s')
print(f'Total time: {elapsed:.3f}s')
print(f'Tokens generated: {token_count}')
print(f'Throughput: {tps:.1f} tokens/s')
````

## OpenAI SDK via Vertex AI

One of vLLM's key selling points: the standard **OpenAI Python SDK works unchanged** against a Vertex AI endpoint. Just point the `base_url` at your dedicated endpoint DNS and use a GCP auth token as the API key.

This means any application built for the OpenAI API — chat interfaces, RAG pipelines, agent frameworks — can switch to your self-hosted vLLM endpoint with a one-line configuration change.

> **Auth token expiry:** GCP access tokens expire after ~1 hour. For production, implement a token refresh mechanism or use a service account with automatic credential rotation.

````python
# Refresh credentials for fresh token
credentials.refresh(google.auth.transport.requests.Request())

# Point the OpenAI SDK at your Vertex AI endpoint
vertex_client = OpenAI(
    base_url=f'https://{endpoint_dns}/v1',
    api_key=credentials.token,  # Use GCP auth token as API key
)

response = vertex_client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {'role': 'system', 'content': 'You are a helpful assistant.'},
        {'role': 'user', 'content': 'What are the key advantages of Mixture of Experts models?'},
    ],
    max_tokens=300,
    temperature=0.7,
)

print(response.choices[0].message.content)
````

### Streaming with OpenAI SDK

The OpenAI SDK's streaming interface also works seamlessly — no need to manually parse SSE.

````python
# Refresh credentials
credentials.refresh(google.auth.transport.requests.Request())
vertex_client = OpenAI(
    base_url=f'https://{endpoint_dns}/v1',
    api_key=credentials.token,
)

stream = vertex_client.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {'role': 'user', 'content': 'Write a haiku about cloud computing.'}
    ],
    max_tokens=100,
    temperature=0.9,
    stream=True,
)

for chunk in stream:
    content = chunk.choices[0].delta.content
    if content:
        print(content, end='', flush=True)

print()
````

## Throughput Testing

Send concurrent requests to measure throughput on managed infrastructure. This tests how the endpoint handles multiple simultaneous users — a key production concern.

We use `ThreadPoolExecutor` to send requests in parallel and measure aggregate throughput (total tokens generated per second across all requests).

````python
def send_request(prompt, request_id):
    """Send a single chat completion request and return timing info."""
    credentials_local, _ = google.auth.default()
    credentials_local.refresh(google.auth.transport.requests.Request())

    payload = {
        'model': MODEL_ID,
        'messages': [{'role': 'user', 'content': prompt}],
        'max_tokens': 200,
        'temperature': 0.7,
    }

    start = time.time()
    resp = httpx.post(
        url,
        headers={
            'Authorization': f'Bearer {credentials_local.token}',
            'Content-Type': 'application/json',
        },
        json=payload,
        timeout=120,
    )
    elapsed = time.time() - start
    result = resp.json()

    completion_tokens = result.get('usage', {}).get('completion_tokens', 0)
    prompt_tokens = result.get('usage', {}).get('prompt_tokens', 0)

    return {
        'request_id': request_id,
        'elapsed': elapsed,
        'prompt_tokens': prompt_tokens,
        'completion_tokens': completion_tokens,
        'tokens_per_sec': completion_tokens / elapsed if elapsed > 0 else 0,
    }
````

````python
prompts = [
    'Explain gradient descent in simple terms.',
    'What is the difference between a compiler and an interpreter?',
    'Describe the CAP theorem in distributed systems.',
    'What are the trade-offs between SQL and NoSQL databases?',
    'Explain how attention mechanisms work in transformers.',
    'What is the difference between horizontal and vertical scaling?',
    'Describe the concept of eventual consistency.',
    'What are the benefits of containerization?',
]

CONCURRENCY = len(prompts)
print(f'Sending {CONCURRENCY} concurrent requests ...\n')

batch_start = time.time()
results = []

with ThreadPoolExecutor(max_workers=CONCURRENCY) as executor:
    futures = {
        executor.submit(send_request, prompt, i): i
        for i, prompt in enumerate(prompts)
    }
    for future in as_completed(futures):
        result = future.result()
        results.append(result)
        print(f'  Request {result["request_id"]}: {result["completion_tokens"]} tokens in {result["elapsed"]:.2f}s ({result["tokens_per_sec"]:.1f} tok/s)')

batch_elapsed = time.time() - batch_start

total_completion_tokens = sum(r['completion_tokens'] for r in results)
total_prompt_tokens = sum(r['prompt_tokens'] for r in results)
avg_latency = sum(r['elapsed'] for r in results) / len(results)

print(f'\n--- Throughput Summary ---')
print(f'Concurrent requests: {CONCURRENCY}')
print(f'Wall-clock time: {batch_elapsed:.2f}s')
print(f'Total prompt tokens: {total_prompt_tokens}')
print(f'Total completion tokens: {total_completion_tokens}')
print(f'Aggregate throughput: {total_completion_tokens / batch_elapsed:.1f} tokens/s')
print(f'Average per-request latency: {avg_latency:.2f}s')
````

- **Specific vLLM version** — pin to a tested version rather than `latest`

## Vertex AI Endpoints vs Other Platforms

| Aspect | Vertex AI Endpoints | Cloud Run | GKE |
|--------|:---:|:---:|:---:|
| **Managed infra** | Fully managed | Fully managed | Self-managed (Autopilot helps) |
| **GPU options** | L4, A100, H100, TPU | L4 only | L4, A100, H100, multi-GPU |
| **Scale-to-zero** | No (min 1 replica) | Yes | Yes (with config) |
| **Prediction API** | `rawPredict` required | Direct HTTP | Direct HTTP |
| **Auth** | IAM + Bearer token | IAM + OIDC | IAM + K8s RBAC |
| **Pre-built container** | Yes (Google optimized) | No | No |
| **Monitoring** | Vertex AI dashboards | Cloud Run metrics | Prometheus + vLLM metrics |
| **Best for** | Simplest managed path | Cost-sensitive, bursty | Full control, high throughput |

Vertex AI Endpoints is the right choice when you want **managed infrastructure** with **minimal operational overhead** and are already using Vertex AI for your ML workflow.
