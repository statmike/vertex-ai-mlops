Deploy vLLM to [GKE Autopilot](https://cloud.google.com/kubernetes-engine/docs/concepts/autopilot-overview) for LLM serving with full Kubernetes control. - Platform comparison: Vertex AI vs Cloud Run vs GKE

**Model:** [google/gemma-4-26B-A4B-it](https://huggingface.co/google/gemma-4-26B-A4B-it) — 26B MoE at INT4, single L4.

## Create GKE Autopilot Cluster

Create a GKE Autopilot cluster. Autopilot manages nodes automatically — you declare the resources your pods need (including GPUs) and GKE provisions the right nodes.

> **Note:** Cluster creation takes ~5-10 minutes. We use the RAPID release channel for the latest Kubernetes features.

````python
cluster_path = f'projects/{PROJECT_ID}/locations/{REGION}'

# Check if cluster already exists
existing = None
try:
    existing = cluster_manager.get_cluster(name=f'{cluster_path}/clusters/{CLUSTER_NAME}')
    print(f'Cluster already exists: {existing.name}')
except Exception:
    pass

if not existing:
    print(f'Creating Autopilot cluster: {CLUSTER_NAME} ...')
    cluster = container_v1.Cluster(
        name=CLUSTER_NAME,
        autopilot=container_v1.Autopilot(enabled=True),
        initial_cluster_version='latest',
        release_channel=container_v1.ReleaseChannel(
            channel=container_v1.ReleaseChannel.Channel.RAPID,
        ),
    )
    operation = cluster_manager.create_cluster(
        request=container_v1.CreateClusterRequest(
            parent=cluster_path,
            cluster=cluster,
        )
    )

    # Wait for cluster creation
    while True:
        op_status = cluster_manager.get_operation(
            name=f'{cluster_path}/operations/{operation.name}'
        )
        if op_status.status == container_v1.Operation.Status.DONE:
            if op_status.error and op_status.error.code:
                raise RuntimeError(f'Cluster creation failed: {op_status.error.message}')
            print('Cluster created!')
            break
        if op_status.status == container_v1.Operation.Status.ABORTING:
            raise RuntimeError(f'Cluster creation aborted: {op_status.status_message}')
        print(f'  Status: {op_status.status.name} ...')
        time.sleep(30)
````

## Configure kubectl

Authenticate to the cluster and configure the Python Kubernetes client.

````python
# Get cluster info
cluster_info = cluster_manager.get_cluster(
    name=f'{cluster_path}/clusters/{CLUSTER_NAME}'
)
print(f'Cluster endpoint: {cluster_info.endpoint}')

# Configure kubernetes client with OAuth2 token
import google.auth
import google.auth.transport.requests
import tempfile

credentials, _ = google.auth.default()
credentials.refresh(google.auth.transport.requests.Request())

# Write cluster CA cert to temp file
ca_cert = base64.b64decode(cluster_info.master_auth.cluster_ca_certificate)
ca_file = tempfile.NamedTemporaryFile(delete=False, suffix='.crt')
ca_file.write(ca_cert)
ca_file.close()

configuration = k8s_client.Configuration()
configuration.host = f'https://{cluster_info.endpoint}'
configuration.ssl_ca_cert = ca_file.name
configuration.api_key = {'authorization': f'Bearer {credentials.token}'}

api_client = k8s_client.ApiClient(configuration)
v1 = k8s_client.CoreV1Api(api_client)
apps_v1 = k8s_client.AppsV1Api(api_client)
print('Kubernetes client configured.')

def refresh_k8s_auth():
    """Refresh the OAuth2 token if it has expired (tokens last ~60 min)."""
    credentials.refresh(google.auth.transport.requests.Request())
    configuration.api_key = {'authorization': f'Bearer {credentials.token}'}
    print('K8s auth token refreshed.')
````

## Create HuggingFace Secret

Store the HuggingFace token as a Kubernetes Secret. The vLLM container references this secret to download gated models like Gemma.

> **Note:** Set the `HF_TOKEN` environment variable before running this cell, or authenticate via `huggingface-cli login`.

````python
NAMESPACE = 'vllm'

# Create namespace
try:
    v1.create_namespace(
        k8s_client.V1Namespace(metadata=k8s_client.V1ObjectMeta(name=NAMESPACE))
    )
    print(f'Namespace {NAMESPACE} created.')
except Exception:
    print(f'Namespace {NAMESPACE} already exists.')

# Get HF token from environment or huggingface_hub cache
hf_token = os.environ.get('HF_TOKEN', '')
if not hf_token:
    try:
        from huggingface_hub import HfFolder
        hf_token = HfFolder.get_token() or ''
    except Exception:
        pass
if not hf_token:
    print('WARNING: No HF_TOKEN found. Set it via environment variable or huggingface-cli login.')
else:
    print(f'HF token found ({len(hf_token)} chars).')

# Create HF token secret
secret = k8s_client.V1Secret(
    metadata=k8s_client.V1ObjectMeta(name='hf-secret', namespace=NAMESPACE),
    type='Opaque',
    data={'token': base64.b64encode(hf_token.encode()).decode()},
)
try:
    v1.create_namespaced_secret(namespace=NAMESPACE, body=secret)
    print('HuggingFace secret created.')
except k8s_client.exceptions.ApiException as e:
    if e.status == 409:
        v1.replace_namespaced_secret(name='hf-secret', namespace=NAMESPACE, body=secret)
        print('HuggingFace secret updated.')
    else:
        raise
````

## Kubernetes Deployment

Deploy vLLM using the official `vllm/vllm-openai` container image. The deployment:
- Requests a single NVIDIA L4 GPU via `nvidia.com/gpu` resource limits
- Uses the `cloud.google.com/gke-accelerator` node selector so Autopilot provisions the right GPU node
- Mounts the HuggingFace token from the Kubernetes Secret
- Uses `bitsandbytes` INT4 quantization to fit Gemma 4 26B-A4B on a single L4 (24 GB)
- Includes readiness and liveness probes on vLLM's `/health` endpoint

````python
MODEL_ID = 'google/gemma-4-26B-A4B-it'
DEPLOYMENT_NAME = 'vllm-gemma'

deployment = k8s_client.V1Deployment(
    metadata=k8s_client.V1ObjectMeta(name=DEPLOYMENT_NAME, namespace=NAMESPACE),
    spec=k8s_client.V1DeploymentSpec(
        replicas=1,
        selector=k8s_client.V1LabelSelector(
            match_labels={'app': 'vllm-gemma'},
        ),
        template=k8s_client.V1PodTemplateSpec(
            metadata=k8s_client.V1ObjectMeta(labels={'app': 'vllm-gemma'}),
            spec=k8s_client.V1PodSpec(
                containers=[
                    k8s_client.V1Container(
                        name='vllm',
                        image='vllm/vllm-openai:latest',
                        args=[
                            '--model', MODEL_ID,
                            '--port', '8000',
                            '--max-model-len', '4096',
                            '--gpu-memory-utilization', '0.90',
                            '--quantization', 'bitsandbytes',
                            '--load-format', 'bitsandbytes',
                            '--trust-remote-code',
                        ],
                        ports=[k8s_client.V1ContainerPort(container_port=8000)],
                        env=[
                            k8s_client.V1EnvVar(
                                name='HF_TOKEN',
                                value_from=k8s_client.V1EnvVarSource(
                                    secret_key_ref=k8s_client.V1SecretKeySelector(
                                        name='hf-secret',
                                        key='token',
                                    ),
                                ),
                            ),
                        ],
                        resources=k8s_client.V1ResourceRequirements(
                            limits={'nvidia.com/gpu': '1'},
                            requests={'cpu': '4', 'memory': '16Gi', 'nvidia.com/gpu': '1'},
                        ),
                        readiness_probe=k8s_client.V1Probe(
                            http_get=k8s_client.V1HTTPGetAction(path='/health', port=8000),
                            initial_delay_seconds=60,
                            period_seconds=10,
                        ),
                        liveness_probe=k8s_client.V1Probe(
                            http_get=k8s_client.V1HTTPGetAction(path='/health', port=8000),
                            initial_delay_seconds=120,
                            period_seconds=30,
                        ),
                    ),
                ],
                node_selector={'cloud.google.com/gke-accelerator': 'nvidia-l4'},
            ),
        ),
    ),
)

try:
    apps_v1.delete_namespaced_deployment(name=DEPLOYMENT_NAME, namespace=NAMESPACE)
    time.sleep(5)
except Exception:
    pass

apps_v1.create_namespaced_deployment(namespace=NAMESPACE, body=deployment)
print(f'Deployment {DEPLOYMENT_NAME} created.')
````

## Create LoadBalancer Service

Expose vLLM via a LoadBalancer Service. This provisions a Google Cloud Network Load Balancer with an external IP.

````python
SERVICE_NAME = 'vllm-service'

service = k8s_client.V1Service(
    metadata=k8s_client.V1ObjectMeta(name=SERVICE_NAME, namespace=NAMESPACE),
    spec=k8s_client.V1ServiceSpec(
        type='LoadBalancer',
        selector={'app': 'vllm-gemma'},
        ports=[k8s_client.V1ServicePort(port=8000, target_port=8000)],
    ),
)

try:
    v1.delete_namespaced_service(name=SERVICE_NAME, namespace=NAMESPACE)
    # Wait for deletion to propagate
    for _ in range(12):
        try:
            v1.read_namespaced_service(SERVICE_NAME, NAMESPACE)
            time.sleep(5)
        except Exception:
            break
except Exception:
    pass

try:
    v1.create_namespaced_service(namespace=NAMESPACE, body=service)
    print(f'Service {SERVICE_NAME} created.')
except Exception as e:
    if '409' in str(e) or 'AlreadyExists' in str(e):
        print(f'Service {SERVICE_NAME} already exists — reusing.')
    else:
        raise
````

## Wait for Deployment Ready

Wait for the vLLM pod to be running and the LoadBalancer to be routable. GPU node provisioning in Autopilot can take 5-15 minutes, and the vLLM container needs additional time to download and load the model.

````python
# Wait for pod to be ready
print('Waiting for vLLM pod to be ready ...')
print('(GPU node provisioning + model download — this can take 10-20 min)')
for i in range(120):
    try:
        pods = v1.list_namespaced_pod(
            namespace=NAMESPACE, label_selector='app=vllm-gemma'
        )
        if pods.items:
            pod = pods.items[0]
            phase = pod.status.phase
            ready = all(
                c.ready for c in (pod.status.container_statuses or [])
                if c.ready is not None
            )
            if phase == 'Running' and ready:
                print(f'Pod is ready! (took ~{i * 15}s)')
                break
            print(f'  Pod status: {phase}, ready: {ready} ({i * 15}s)')
        else:
            print(f'  No pods yet ... ({i * 15}s)')
    except Exception as e:
        if '401' in str(e):
            refresh_k8s_auth()
        else:
            print(f'  Waiting ... ({i * 15}s)')
    time.sleep(15)
else:
    print('Timeout waiting for pod. Check pod events with:')
    print(f'  kubectl describe pods -n {NAMESPACE} -l app=vllm-gemma')
````

````python
# Get external IP
print('Waiting for external IP ...')
EXTERNAL_IP = None
for i in range(30):
    try:
        svc = v1.read_namespaced_service(name=SERVICE_NAME, namespace=NAMESPACE)
        ingress = svc.status.load_balancer.ingress
        if ingress and ingress[0].ip:
            EXTERNAL_IP = ingress[0].ip
            print(f'External IP: {EXTERNAL_IP}')
            break
    except Exception as e:
        if '401' in str(e):
            refresh_k8s_auth()
    time.sleep(10)
else:
    print('Timeout waiting for external IP.')

# Wait for LoadBalancer to be routable (IP assigned != traffic flowing)
if EXTERNAL_IP:
    print('\nWaiting for LoadBalancer to accept traffic ...')
    for i in range(30):
        try:
            resp = httpx.get(f'http://{EXTERNAL_IP}:8000/health', timeout=5)
            if resp.status_code == 200:
                print('vLLM is reachable and ready!')
                break
        except Exception:
            pass
        if i % 3 == 0:
            print(f'  Not yet routable ... ({i * 10}s)')
        time.sleep(10)
    else:
        print('Timeout waiting for LoadBalancer routing.')
````

## Inference via OpenAI SDK

vLLM exposes an OpenAI-compatible API. Use the standard OpenAI Python SDK pointed at the LoadBalancer IP — no authentication needed (unlike Vertex AI or Cloud Run).

````python
client_oai = OpenAI(
    base_url=f'http://{EXTERNAL_IP}:8000/v1',
    api_key='unused',
)

response = client_oai.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {'role': 'user', 'content': 'What are the advantages of running LLMs on Kubernetes?'}
    ],
    max_tokens=256,
    temperature=0.7,
)
print(response.choices[0].message.content)
````

### Streaming Inference

vLLM supports server-sent events (SSE) for streaming token generation. This is especially useful for interactive applications where you want to display tokens as they are generated.

````python
stream = client_oai.chat.completions.create(
    model=MODEL_ID,
    messages=[
        {'role': 'user', 'content': 'Explain tensor parallelism for LLM serving in 3 sentences.'}
    ],
    max_tokens=256,
    temperature=0.7,
    stream=True,
)

print('Streaming response:')
for chunk in stream:
    if chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end='', flush=True)
print()
````

## HPA with vLLM Metrics

vLLM exposes [Prometheus metrics](https://docs.vllm.ai/en/latest/serving/metrics.html) on its `/metrics` endpoint. These are LLM-specific signals that make much better autoscaling targets than generic CPU/memory:

| Metric | Description | HPA Use |
|--------|-------------|----------|
| `vllm:num_requests_running` | Current in-flight requests | Scale on active load |
| `vllm:num_requests_waiting` | Queued requests | Scale on backpressure |
| `vllm:gpu_cache_usage_perc` | GPU KV cache utilization | Scale before OOM |
| `vllm:avg_generation_throughput_toks_per_s` | Token throughput | Monitor capacity |

Below is the HPA configuration pattern using `vllm:num_requests_running`. This scales up when the average number of in-flight requests per pod exceeds 5.

> **Note:** Using custom Prometheus metrics with HPA requires a [metrics pipeline](https://cloud.google.com/kubernetes-engine/docs/how-to/custom-metrics) — Prometheus adapter or GKE Managed Prometheus with the Custom Metrics Stackdriver Adapter. The manifest below shows the HPA configuration; the metrics pipeline setup is a one-time cluster configuration.

````python
hpa_manifest = {
    'apiVersion': 'autoscaling/v2',
    'kind': 'HorizontalPodAutoscaler',
    'metadata': {'name': 'vllm-hpa', 'namespace': NAMESPACE},
    'spec': {
        'scaleTargetRef': {
            'apiVersion': 'apps/v1',
            'kind': 'Deployment',
            'name': DEPLOYMENT_NAME,
        },
        'minReplicas': 1,
        'maxReplicas': 3,
        'metrics': [
            {
                'type': 'Pods',
                'pods': {
                    'metric': {'name': 'vllm:num_requests_running'},
                    'target': {'type': 'AverageValue', 'averageValue': '5'},
                },
            },
        ],
    },
}

# Print the HPA manifest for reference
print(yaml.dump(hpa_manifest, default_flow_style=False))
````

For comparison, here is a simpler HPA using the `vllm:gpu_cache_usage_perc` metric — this scales based on GPU memory pressure, which is the primary bottleneck for LLM serving:

```yaml
apiVersion: autoscaling/v2
kind: HorizontalPodAutoscaler
metadata:
  name: vllm-hpa-cache
  namespace: vllm
spec:
  scaleTargetRef:
    apiVersion: apps/v1
    kind: Deployment
    name: vllm-gemma
  minReplicas: 1
  maxReplicas: 5
  metrics:
    - type: Pods
      pods:
        metric:
          name: vllm:gpu_cache_usage_perc
        target:
          type: AverageValue
          averageValue: "80"     # Scale up when KV cache > 80%
```

Setting up the custom metrics pipeline requires:
1. [GKE Managed Prometheus](https://cloud.google.com/stackdriver/docs/managed-prometheus) to scrape vLLM's `/metrics` endpoint
2. [Custom Metrics Adapter](https://cloud.google.com/kubernetes-engine/docs/how-to/custom-metrics) to expose those metrics to the HPA controller

## GKE Inference Gateway

The [GKE Inference Gateway](https://cloud.google.com/kubernetes-engine/docs/concepts/inference-gateway) is purpose-built for LLM workloads on GKE. Unlike a standard LoadBalancer (which routes round-robin), the Inference Gateway understands LLM-specific signals:

- **Queue-depth-aware routing:** Routes requests to the replica with the shortest queue, reducing latency
- **KV cache utilization:** Avoids sending requests to pods with high GPU memory pressure
- **Model-aware load balancing:** Understands that LLM requests have variable cost (long prompts vs short ones)
- **Works with vLLM metrics:** Reads directly from vLLM's Prometheus `/metrics` endpoint

The Inference Gateway replaces the simple LoadBalancer Service with an intelligent Gateway resource:

```yaml
apiVersion: inference.networking.x-k8s.io/v1alpha2
kind: InferencePool
metadata:
  name: vllm-pool
spec:
  targetPortNumber: 8000
  selector:
    app: vllm-gemma
  extensionRef:
    name: endpoint-picker
---
apiVersion: inference.networking.x-k8s.io/v1alpha2
kind: InferenceModel
metadata:
  name: gemma-model
spec:
  modelName: google/gemma-4-26B-A4B-it
  criticality: Critical
  poolRef:
    name: vllm-pool
```

This is a key GKE advantage over Cloud Run and Vertex AI — intelligent, LLM-aware request routing at the infrastructure level.

> **Learn more:** [GKE Inference Gateway documentation](https://cloud.google.com/kubernetes-engine/docs/how-to/serve-llm-with-inference-gateway)

## Multi-GPU Deployment

For models that don't fit on a single GPU (e.g., large dense models at BF16), vLLM supports **tensor parallelism** to split the model across multiple GPUs.

Key changes for multi-GPU:

1. **`--tensor-parallel-size N`** — Splits model layers across N GPUs
2. **`nvidia.com/gpu: 'N'`** — Request N GPUs in the pod resource limits
3. **Node selector** — Use a multi-GPU node type (e.g., `nvidia-a100-80gb` with 2 or 8 GPUs)

```python
# Example: 2-GPU deployment for a larger model
args=[
    '--model', 'google/gemma-3-27b-it',
    '--port', '8000',
    '--tensor-parallel-size', '2',
    '--max-model-len', '8192',
    '--gpu-memory-utilization', '0.90',
],
resources=k8s_client.V1ResourceRequirements(
    limits={'nvidia.com/gpu': '2'},
    requests={'cpu': '8', 'memory': '32Gi', 'nvidia.com/gpu': '2'},
),
node_selector={'cloud.google.com/gke-accelerator': 'nvidia-a100-80gb'},
```

GKE also supports **multi-node** tensor parallelism via [LeaderWorkerSet](https://cloud.google.com/kubernetes-engine/docs/how-to/multislice-gpus) for models that don't fit on a single machine. This is unique to GKE — neither Cloud Run nor Vertex AI support multi-node GPU deployments.

## Monitoring

vLLM exposes Prometheus metrics on `/metrics`. These provide deep visibility into the serving engine's performance — request queues, KV cache utilization, token throughput, and more.

````python
resp = httpx.get(f'http://{EXTERNAL_IP}:8000/metrics', timeout=10)

# Parse and display key vLLM metrics
key_patterns = [
    'num_requests_running',
    'num_requests_waiting',
    'gpu_cache_usage',
    'generation_throughput',
    'num_preemptions',
    'prompt_tokens_total',
    'generation_tokens_total',
]

print('vLLM Metrics (selected):\n')
for line in resp.text.split('\n'):
    if any(m in line for m in key_patterns) and not line.startswith('#'):
        print(f'  {line}')
````
