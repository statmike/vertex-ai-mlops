## When to Use Triton + vLLM

Most users should use standalone vLLM. Adding Triton introduces an extra layer — more configuration, a larger container image, and another component to debug. The Triton + vLLM combination is for teams that already have Triton infrastructure or need capabilities that standalone vLLM does not provide.

| Scenario | Recommendation |
|----------|---------------|
| Serve only LLMs | Standalone vLLM |
| Serve LLMs + traditional ML models from one endpoint | Triton + vLLM backend |
| Need ensemble pipelines (preprocessing -> LLM -> postprocessing) | Triton + vLLM backend |
| Need KServe V2 protocol for existing infrastructure | Triton + vLLM backend |
| Need multi-model routing via Triton's model management | Triton + vLLM backend |
| Simplest deployment | Standalone vLLM |
| Google Cloud managed (Vertex AI) | Standalone vLLM (pre-built container) |

The key value proposition is **multi-model serving**: one Triton server can host an LLM (via the vLLM backend) alongside traditional ML models (ONNX, TensorFlow, PyTorch) with a unified API. Standalone vLLM can only serve LLMs.

---
## How the vLLM Backend Works

NVIDIA provides an [official vLLM backend](https://github.com/triton-inference-server/vllm_backend) for Triton Inference Server. Here is what happens under the hood:

1. **Wrapping vLLM's engine:** The backend instantiates vLLM's `AsyncLLMEngine` inside a Triton Python backend model. All of vLLM's optimizations — PagedAttention, continuous batching, prefix caching — run inside this engine.

2. **Request translation:** Incoming requests arrive via Triton's KServe V2 protocol (typed tensors like `text_input`, `max_tokens`, `temperature`). The backend translates these into vLLM's internal `SamplingParams` and prompt format.

3. **vLLM handles inference:** vLLM manages GPU memory, schedules sequences, and generates tokens. Triton does not interfere with vLLM's batching or scheduling — it delegates entirely.

4. **Triton handles the rest:** Model lifecycle (load/unload), health checks, Prometheus metrics, multi-model routing, and the OpenAI-compatible API endpoint are all managed by Triton.

5. **Dual API surface:** The backend exposes both Triton's standard KServe V2 endpoints *and* an OpenAI-compatible `/v1/chat/completions` endpoint. Clients can choose whichever protocol fits their infrastructure.

```
Client Request
     |
     v
Triton Inference Server
  |-- KServe V2 API (/v2/models/gemma_llm/infer)
  |-- OpenAI API   (/v1/chat/completions)
  |
  v
vLLM Backend (Python backend wrapper)
  |
  v
vLLM AsyncLLMEngine
  |-- PagedAttention
  |-- Continuous batching
  |-- GPU memory management
```

For details on how vLLM's engine works, see the [vLLM Fundamentals](./vLLM%20-%20Fundamentals.ipynb) notebook. For Triton's model repository and config concepts, see the [Triton Fundamentals](../Triton/Triton%20Inference%20Server%20-%20Fundamentals.ipynb) notebook.

## Model Repository Setup

Create the Triton model repository with a vLLM backend configuration. The directory structure follows the standard Triton layout — a model directory with a `config.pbtxt` and a numbered version subdirectory.

```
model_repository/
  gemma_llm/
    config.pbtxt        <- vLLM backend configuration
    1/
      .keep             <- empty version directory (config defines everything)
```

Unlike ONNX or Python backend models where the version directory contains model artifacts, the vLLM backend downloads the model from HuggingFace at load time. The version directory is empty — similar to Triton ensemble models.

````python
# Create model repository structure
vllm_model_dir = f'{MODEL_REPO}/gemma_llm/1'
os.makedirs(vllm_model_dir, exist_ok=True)
````

````python
config_pbtxt = '''name: "gemma_llm"
backend: "vllm"

# Model configuration for vLLM backend
parameters: {
  key: "model"
  value: {
    string_value: "google/gemma-4-E2B-it"
  }
}
parameters: {
  key: "gpu_memory_utilization"
  value: {
    string_value: "0.85"
  }
}
parameters: {
  key: "max_model_len"
  value: {
    string_value: "4096"
  }
}
parameters: {
  key: "dtype"
  value: {
    string_value: "auto"
  }
}
parameters: {
  key: "trust_remote_code"
  value: {
    string_value: "true"
  }
}

instance_group [
  {
    count: 1
    kind: KIND_MODEL
  }
]
'''

with open(f'{MODEL_REPO}/gemma_llm/config.pbtxt', 'w') as f:
    f.write(config_pbtxt)
print(f'Wrote {MODEL_REPO}/gemma_llm/config.pbtxt')
````

**Configuration explained:**

- `backend: "vllm"` tells Triton to use the vLLM backend (included in the `tritonserver:*-vllm-python-py3` image)
- Parameters map directly to vLLM engine arguments (the same flags you would pass to `vllm serve` on the CLI)
- `instance_group` with `KIND_MODEL` means vLLM manages its own GPU allocation — Triton does not assign specific GPUs
- `gpu_memory_utilization: 0.85` reserves 85% of GPU memory for KV cache (same as vLLM's default)
- `max_model_len: 4096` caps the maximum sequence length to keep memory usage predictable

````python
# vLLM backend models have empty version directories (config defines everything)
# Place a .keep file so the directory survives git/GCS operations
with open(f'{vllm_model_dir}/.keep', 'w') as f:
    pass

print(f'Model repository structure:')
for root, dirs, files in sorted(os.walk(MODEL_REPO)):
    rel = os.path.relpath(root, MODEL_REPO)
    level = rel.count(os.sep)
    indent = '  ' * level
    dirname = os.path.basename(root)
    if rel == '.':
        print('model_repository/')
    else:
        print(f'{indent}{dirname}/')
    subindent = '  ' * (level + 1)
    for f in sorted(files):
        print(f'{subindent}{f}')
````

## Build and Start Container

````python
docker_client = docker.from_env()

# Build the image
print('Building Triton + vLLM container ...')
image, build_logs = docker_client.images.build(
    path=DIR,
    tag='triton-vllm:latest',
    rm=True,
)
for chunk in build_logs:
    if 'stream' in chunk:
        print(chunk['stream'], end='')
print('Build complete.')
````

````python
# Stop any existing container with the same name
try:
    old = docker_client.containers.get('triton-vllm')
    old.stop(timeout=10)
    old.remove()
    print('Removed existing triton-vllm container.')
except docker.errors.NotFound:
    pass

# Start the container
container = docker_client.containers.run(
    'triton-vllm:latest',
    detach=True,
    name='triton-vllm',
    ports={'8000/tcp': 8000},
    runtime='nvidia',  # GPU access
    environment={
        'HF_TOKEN': os.environ.get('HF_TOKEN', ''),
    },
    shm_size='2g',  # Shared memory for Python backend
    remove=True,
)
print(f'Container started: {container.id[:12]}')
````

````python
print('Waiting for Triton + vLLM to be ready ...')
for i in range(120):
    try:
        resp = httpx.get('http://localhost:8000/v2/health/ready', timeout=2)
        if resp.status_code == 200:
            print(f'Triton is ready! (took ~{i * 5}s)')
            break
    except Exception:
        pass
    time.sleep(5)
else:
    print('WARNING: Server did not become ready. Check container logs:')
    print(container.logs().decode()[-2000:])
````

## Inference via Triton Client (KServe V2)

The vLLM backend accepts requests via Triton's standard KServe V2 protocol. Input and output tensors use the names defined by the backend:

- **Input:** `text_input` (BYTES) — the prompt text
- **Input (optional):** `max_tokens` (INT32), `temperature` (FP32), etc.
- **Output:** `text_output` (BYTES) — the generated text

This is the same `tritonclient` library used in the [Triton Fundamentals](../Triton/Triton%20Inference%20Server%20-%20Fundamentals.ipynb) notebook — the only difference is the model name and tensor names.

````python
triton_client = httpclient.InferenceServerClient(url='localhost:8000')

# Check model is ready
assert triton_client.is_model_ready('gemma_llm'), 'Model not ready!'

# The vLLM backend accepts text_input and returns text_output
prompt = 'Explain why Triton and vLLM complement each other for model serving.'

# Create input tensor
input_text = httpclient.InferInput('text_input', [1], 'BYTES')
input_text.set_data_from_numpy(np.array([prompt], dtype=object))

# Optional: set generation parameters
max_tokens = httpclient.InferInput('max_tokens', [1], 'INT32')
max_tokens.set_data_from_numpy(np.array([256], dtype=np.int32))

temperature = httpclient.InferInput('temperature', [1], 'FP32')
temperature.set_data_from_numpy(np.array([0.7], dtype=np.float32))

# Request output
output = httpclient.InferRequestedOutput('text_output')

# Send request
result = triton_client.infer(
    model_name='gemma_llm',
    inputs=[input_text, max_tokens, temperature],
    outputs=[output],
)

response_text = result.as_numpy('text_output')[0].decode()
print(response_text)
````

**KServe V2 vs OpenAI protocol:**

| Aspect | KServe V2 (`tritonclient`) | OpenAI SDK |
|--------|---------------------------|------------|
| Input format | Typed tensors (`BYTES`, `INT32`, `FP32`) | JSON messages (`{"role": ..., "content": ...}`) |
| Output format | Named tensor (`text_output`) | JSON with `choices[0].message.content` |
| Protocol | Triton-native, tensor-oriented | REST/JSON, chat-oriented |
| Best for | Existing Triton infrastructure, pipelines | Drop-in replacement for OpenAI API clients |
| Streaming | Triton's decoupled mode | Standard SSE |

Both hit the same vLLM engine underneath. Choose based on your client infrastructure.

## Inference via OpenAI SDK

The vLLM backend in Triton also exposes an OpenAI-compatible `/v1/chat/completions` endpoint alongside the standard KServe V2 endpoints. This means existing code written for the OpenAI API works without changes — just point it at the Triton server.

````python
# The Triton vLLM backend exposes OpenAI-compatible endpoints
# alongside the standard KServe V2 endpoints
oai_client = OpenAI(base_url='http://localhost:8000/v1', api_key='unused')

response = oai_client.chat.completions.create(
    model='gemma_llm',  # Use the Triton model name
    messages=[
        {'role': 'user', 'content': 'What is the advantage of serving an LLM through Triton instead of standalone vLLM?'}
    ],
    max_tokens=256,
    temperature=0.7,
)
print(response.choices[0].message.content)
````

## Multi-Model Serving

This is the key advantage of using Triton with the vLLM backend: **one Triton server can host both LLMs and traditional ML models** with a unified API. Standalone vLLM can only serve LLMs.

| Model | Backend | Use Case |
|-------|---------|----------|
| `sentiment_onnx` | ONNX Runtime | Text classification (traditional ML) |
| `gemma_llm` | vLLM | Text generation (LLM) |

In a production scenario, you would add ONNX, TensorFlow, or PyTorch models to the same model repository. One Triton server handles all inference — traditional ML and LLM — with clients routing to the right model by name.

For example:
- A **sentiment classifier** (ONNX, <100ms latency) processes incoming customer messages
- Negative sentiment messages get routed to a **Gemma LLM** (vLLM backend) for empathetic response generation
- Both models run on the same Triton instance, sharing GPU resources

This can also be composed into a Triton ensemble pipeline — see the [Triton Pipelines and Ensembles](../Triton/Triton%20Inference%20Server%20-%20Pipelines%20and%20Ensembles.ipynb) notebook for how to chain models declaratively.

See the [Triton Fundamentals](../Triton/Triton%20Inference%20Server%20-%20Fundamentals.ipynb) notebook for how to add ONNX models to the model repository.

## Streaming via Triton

Streaming is supported via the vLLM backend's decoupled mode. The OpenAI-compatible endpoint supports standard SSE (Server-Sent Events) streaming — the same pattern used in the [vLLM Fundamentals](./vLLM%20-%20Fundamentals.ipynb) notebook.

````python
# Streaming is supported via the vLLM backend's decoupled mode
# The OpenAI-compatible endpoint supports standard SSE streaming

stream = oai_client.chat.completions.create(
    model='gemma_llm',
    messages=[
        {'role': 'user', 'content': 'Write a short poem about model serving.'}
    ],
    max_tokens=100,
    stream=True,
)

for chunk in stream:
    if chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end='', flush=True)
print()
````

## When NOT to Use This

Be honest about the trade-offs. Triton + vLLM is not always the right choice:

- **Just serving LLMs?** Use standalone vLLM — simpler, fewer moving parts, smaller container image
- **Using Vertex AI?** Use Google's [pre-built vLLM container](https://cloud.google.com/vertex-ai/docs/predictions/pre-built-containers) — no Triton needed, managed scaling
- **Don't need multi-model?** Standalone vLLM is more straightforward and has a smaller operational surface
- **Triton adds overhead:** Extra layer means more configuration, more debugging surface, and larger container images (~20 GB vs ~5 GB for standalone vLLM)
- **Version coupling:** Triton and vLLM versions must be compatible — the vLLM backend is pinned to the vLLM version included in the Triton image. You cannot independently upgrade vLLM without upgrading the Triton image
- **Community support:** Standalone vLLM has a larger community and faster release cycle. Issues at the Triton-vLLM boundary can be harder to debug
