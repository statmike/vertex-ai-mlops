# Serving Models on GKE

Deploy the same custom prediction container to [GKE Autopilot](https://cloud.google.com/kubernetes-engine/docs/concepts/autopilot-overview) and compare with Cloud Run and Vertex AI Endpoints. Same container, different platform — built once, runs everywhere.

This notebook mirrors the [Cloud Run notebook](Serving%20Models%20on%20Cloud%20Run.ipynb) — same FastAPI sentiment container, deployed to GKE Autopilot instead of Cloud Run. The container portability principle: the same Docker image runs on Vertex AI, Cloud Run, and GKE without code changes, thanks to the `AIP_*` environment variables.

**What you'll learn:**
- Create a GKE Autopilot cluster (fully managed Kubernetes — no node management)
- Deploy the container as a Kubernetes Deployment + Service
- Configure Horizontal Pod Autoscaler (HPA) for CPU-based autoscaling
- Deploy a second model version and split traffic between v1 and v2

**When to choose GKE:**
- Custom networking requirements (VPC-native, Ingress controllers, Gateway API)
- GPU scheduling with fine-grained control (node pools, taints/tolerations, multi-GPU)
- Multi-model pods sharing node resources
- Compliance or security requirements that mandate Kubernetes
- Team already on Kubernetes and wants to keep the ML stack consistent

**Runtime:** ~15-20 minutes (cluster creation takes ~10 minutes).

**Prerequisites:** Familiarity with the [Cloud Run notebook](Serving%20Models%20on%20Cloud%20Run.ipynb) pattern. Same container, different platform.

## Download HuggingFace Models

Download two sentiment analysis models from HuggingFace and upload them to GCS. Both models are fine-tuned on the [SST-2](https://huggingface.co/datasets/stanfordnlp/sst2) dataset for binary sentiment classification.

- **v1 (DistilBERT):** Smaller, faster — labels: `POSITIVE` / `NEGATIVE`
- **v2 (BERT):** Larger, more accurate — labels: `LABEL_1` / `LABEL_0`

The different label formats make it easy to identify which model served a request during traffic splitting.

```python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
```

```python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        # upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    # store the GCS URI
    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
```

```python
for version, info in MODELS.items():
    blobs = list(bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/models/{version}'))
    print(f'{version} ({info["display_name"]}): {len(blobs)} files at {info["artifact_uri"]}')
```

```python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)

for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)

@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}

@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']
    results = classifier(instances)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
```

## GKE Autopilot Cluster

[GKE Autopilot](https://cloud.google.com/kubernetes-engine/docs/concepts/autopilot-overview) is a fully managed Kubernetes mode where Google manages the nodes, scaling, and security. You only define pods — Autopilot provisions the right nodes automatically.

**Key properties:**
- **No node management:** No node pools, no OS patching, no capacity planning
- **Pay per pod:** Billed for the CPU, memory, and GPU requested by your pods (not idle nodes)
- **Auto-scaling:** Nodes scale automatically based on pod demand; combine with HPA for full autoscaling
- **Security by default:** Hardened nodes, Workload Identity, Shielded GKE Nodes

Compared to Cloud Run: GKE Autopilot gives you the full Kubernetes API (Deployments, Services, Ingress, Gateway, ConfigMaps, Secrets, RBAC) while still being fully managed. Cloud Run is simpler but less flexible.

```python
parent = f'projects/{PROJECT_ID}/locations/{REGION}'

# Check for existing cluster
try:
    cluster = cluster_client.get_cluster(name=f'{parent}/clusters/{GKE_CLUSTER}')
    print(f'Cluster already exists: {cluster.name} (status: {cluster.status.name})')
except Exception:
    cluster = container_v1.Cluster(
        name=GKE_CLUSTER,
        autopilot=container_v1.Autopilot(enabled=True),
        release_channel=container_v1.ReleaseChannel(
            channel=container_v1.ReleaseChannel.Channel.REGULAR,
        ),
    )

    operation = cluster_client.create_cluster(
        parent=parent,
        cluster=cluster,
    )
    print(f'Creating GKE Autopilot cluster (this takes ~10 minutes)...')

    # Poll for completion
    while True:
        op = cluster_client.get_operation(name=f'{parent}/operations/{operation.name}')
        if op.status == container_v1.Operation.Status.DONE:
            break
        time.sleep(30)
        print(f'  Status: {op.status.name}...')

    cluster = cluster_client.get_cluster(name=f'{parent}/clusters/{GKE_CLUSTER}')
    print(f'Cluster created: {cluster.name}')
```

### Configure Workload Identity

GKE Autopilot enables [Workload Identity](https://cloud.google.com/kubernetes-engine/docs/how-to/workload-identity) by default — pods do **not** inherit the node's service account credentials. Each Kubernetes service account must be explicitly bound to a GCP service account to access GCP resources like GCS.

We bind the `default` service account in the `serving` namespace to the default compute service account, which has access to the project's GCS buckets.

```python
GCP_SA = f'{PROJECT_NUMBER}-compute@developer.gserviceaccount.com'

# Allow the Kubernetes service account to impersonate the GCP service account
!gcloud iam service-accounts add-iam-policy-binding {GCP_SA} \
    --role roles/iam.workloadIdentityUser \
    --member "serviceAccount:{PROJECT_ID}.svc.id.goog[{NAMESPACE}/default]" \
    --condition=None \
    --quiet

# Annotate the Kubernetes service account with the GCP service account
v1.patch_namespaced_service_account(
    name='default',
    namespace=NAMESPACE,
    body={'metadata': {'annotations': {'iam.gke.io/gcp-service-account': GCP_SA}}},
)
print(f'Workload Identity: {NAMESPACE}/default → {GCP_SA}')
```

## Deploy Model v1

Deploy the DistilBERT model as a Kubernetes Deployment with a LoadBalancer Service. This is the standard Kubernetes pattern:

- **Deployment:** Manages the pods running the container (replicas, rolling updates, rollbacks)
- **Service:** Exposes the pods via a stable network endpoint (LoadBalancer gives us an external IP)

The `AIP_*` environment variables are set explicitly in the container spec — same variables that Vertex AI sets automatically.

```python
AR_IMAGE = f'{REGION}-docker.pkg.dev/{PROJECT_ID}/{AR_REPO}/{EXPERIMENT}:latest'

deployment_v1 = {
    'apiVersion': 'apps/v1',
    'kind': 'Deployment',
    'metadata': {'name': 'sentiment-v1', 'namespace': NAMESPACE},
    'spec': {
        'replicas': 1,
        'selector': {'matchLabels': {'app': 'sentiment', 'version': 'v1'}},
        'template': {
            'metadata': {'labels': {'app': 'sentiment', 'version': 'v1'}},
            'spec': {
                'containers': [{
                    'name': 'sentiment',
                    'image': AR_IMAGE,
                    'ports': [{'containerPort': 8080}],
                    'env': [
                        {'name': 'AIP_STORAGE_URI', 'value': MODELS['v1']['artifact_uri']},
                        {'name': 'AIP_HTTP_PORT', 'value': '8080'},
                        {'name': 'AIP_HEALTH_ROUTE', 'value': '/health'},
                        {'name': 'AIP_PREDICT_ROUTE', 'value': '/predict'},
                        # Limit gunicorn to 1 worker — the base image auto-scales by
                        # node CPU cores, but each worker loads the full model into memory.
                        # On Autopilot, the node may have 16+ cores while the pod only has 4Gi.
                        {'name': 'WEB_CONCURRENCY', 'value': '1'},
                    ],
                    'resources': {
                        'requests': {'cpu': '500m', 'memory': '2Gi'},
                        'limits': {'cpu': '2', 'memory': '4Gi'},
                    },
                    'readinessProbe': {
                        'httpGet': {'path': '/health', 'port': 8080},
                        'initialDelaySeconds': 120,
                        'periodSeconds': 10,
                    },
                    'livenessProbe': {
                        'httpGet': {'path': '/health', 'port': 8080},
                        'initialDelaySeconds': 180,
                        'periodSeconds': 30,
                    },
                }],
            },
        },
    },
}

try:
    apps_v1.create_namespaced_deployment(namespace=NAMESPACE, body=deployment_v1)
    print('Deployment v1 created')
except Exception as e:
    if '409' in str(e):
        apps_v1.replace_namespaced_deployment(name='sentiment-v1', namespace=NAMESPACE, body=deployment_v1)
        print('Deployment v1 updated')
    else:
        raise
```

```python
service = {
    'apiVersion': 'v1',
    'kind': 'Service',
    'metadata': {'name': 'sentiment-service', 'namespace': NAMESPACE},
    'spec': {
        'type': 'LoadBalancer',
        'selector': {'app': 'sentiment', 'version': 'v1'},
        'ports': [{'port': 80, 'targetPort': 8080}],
    },
}

core_v1 = k8s_client.CoreV1Api()
try:
    core_v1.create_namespaced_service(namespace=NAMESPACE, body=service)
    print('Service created')
except Exception as e:
    if '409' in str(e):
        print('Service already exists')
    else:
        raise
```

### Wait for External IP

```python
# Wait for external IP
EXTERNAL_IP = None
for _ in range(30):
    svc = core_v1.read_namespaced_service('sentiment-service', NAMESPACE)
    ingress = svc.status.load_balancer.ingress
    if ingress and ingress[0].ip:
        EXTERNAL_IP = ingress[0].ip
        break
    time.sleep(10)
    print('Waiting for external IP...')

print(f'External IP: {EXTERNAL_IP}')
SERVICE_URL = f'http://{EXTERNAL_IP}'
```

## Deploy Model v2 + Traffic Splitting

Deploy the BERT model (v2) alongside DistilBERT (v1). In Kubernetes, traffic splitting between versions is controlled by **Service selectors** — the Service routes to all pods matching its label selector.

**Strategy:**
1. Deploy v2 as a separate Deployment (same `app` label, different `version` label)
2. Update the Service selector to match both versions (round-robin traffic)
3. Verify both models are serving
4. Switch entirely to v2

For weighted traffic splitting (e.g., 80/20), you would use the [Kubernetes Gateway API](https://gateway-api.sigs.k8s.io/) or [Istio](https://istio.io/) — beyond the scope of this demo.

```python
deployment_v2 = {
    'apiVersion': 'apps/v1',
    'kind': 'Deployment',
    'metadata': {'name': 'sentiment-v2', 'namespace': NAMESPACE},
    'spec': {
        'replicas': 1,
        'selector': {'matchLabels': {'app': 'sentiment', 'version': 'v2'}},
        'template': {
            'metadata': {'labels': {'app': 'sentiment', 'version': 'v2'}},
            'spec': {
                'containers': [{
                    'name': 'sentiment',
                    'image': AR_IMAGE,
                    'ports': [{'containerPort': 8080}],
                    'env': [
                        {'name': 'AIP_STORAGE_URI', 'value': MODELS['v2']['artifact_uri']},
                        {'name': 'AIP_HTTP_PORT', 'value': '8080'},
                        {'name': 'AIP_HEALTH_ROUTE', 'value': '/health'},
                        {'name': 'AIP_PREDICT_ROUTE', 'value': '/predict'},
                        {'name': 'WEB_CONCURRENCY', 'value': '1'},
                    ],
                    'resources': {
                        'requests': {'cpu': '500m', 'memory': '2Gi'},
                        'limits': {'cpu': '2', 'memory': '4Gi'},
                    },
                    'readinessProbe': {
                        'httpGet': {'path': '/health', 'port': 8080},
                        'initialDelaySeconds': 120,
                        'periodSeconds': 10,
                    },
                    'livenessProbe': {
                        'httpGet': {'path': '/health', 'port': 8080},
                        'initialDelaySeconds': 180,
                        'periodSeconds': 30,
                    },
                }],
            },
        },
    },
}

try:
    apps_v1.create_namespaced_deployment(namespace=NAMESPACE, body=deployment_v2)
    print('Deployment v2 created')
except Exception as e:
    if '409' in str(e):
        apps_v1.replace_namespaced_deployment(name='sentiment-v2', namespace=NAMESPACE, body=deployment_v2)
        print('Deployment v2 updated')
    else:
        raise

# Wait for v2 pod to be ready
print('Waiting for v2 pod to be ready...')
for i in range(60):  # up to 10 minutes
    pods = v1.list_namespaced_pod(NAMESPACE, label_selector='app=sentiment,version=v2')
    # Sort by creation timestamp descending to get the newest pod
    active_pods = sorted(pods.items, key=lambda p: p.metadata.creation_timestamp, reverse=True)
    pod = active_pods[0] if active_pods else None
    pod_phase = pod.status.phase if pod else 'No pods'
    container_ready = False
    if pod and pod.status.container_statuses:
        container_ready = all(c.ready for c in pod.status.container_statuses)
    if container_ready:
        print(f'v2 pod is ready! (took ~{i * 10}s)')
        break
    if i % 6 == 0:
        print(f'  {i * 10}s: pod={pod_phase}, container_ready={container_ready}')
    time.sleep(10)
else:
    raise RuntimeError('v2 pod did not become ready within 10 minutes.')
```

### Route Traffic to Both Versions

After updating the Service selector to match both `v1` and `v2` pods, Kubernetes distributes new connections across all matching endpoints. However, GKE's Network Load Balancer routes at layer 4 using source-IP affinity — requests from the same client IP are pinned to the same backend. From a single notebook client, you may see all traffic land on one model. In production with multiple clients, traffic distributes across both versions.

```python
# Update service to route to both versions
service_patch = {
    'spec': {
        'selector': {'app': 'sentiment'},  # remove version label -- matches both v1 and v2
    },
}
core_v1.patch_namespaced_service('sentiment-service', NAMESPACE, body=service_patch)
print('Service now routes to both v1 and v2 (round-robin)')
```

```python
# Send 20 requests and observe traffic distribution
# DistilBERT (v1) uses labels POSITIVE/NEGATIVE, while BERT (v2) uses LABEL_0/LABEL_1.
# This makes it easy to identify which model served each request.
# Note: Connection: close forces a new TCP connection per request — without this,
# Python requests reuses the same connection, which conntrack pins to one backend.
from collections import Counter

print('Sending 20 requests to check traffic distribution...\n')
labels = []
for i in range(20):
    result = requests.post(
        f'{SERVICE_URL}/predict',
        json={'instances': ['This movie is great!']},
        headers={'Connection': 'close'},
        timeout=30,
    ).json()
    label = result['predictions'][0]['label']
    labels.append(label)

label_counts = Counter(labels)
print('Label distribution:')
for label, count in sorted(label_counts.items()):
    print(f'  {label:12s}: {count:2d} requests')

print(f'\n{len(label_counts)} distinct label format(s) observed across 20 requests.')
if len(label_counts) > 1:
    print('Different label formats confirm traffic is split between v1 (POSITIVE/NEGATIVE) and v2 (LABEL_0/LABEL_1).')
```

**Note:** Kubernetes Service selectors provide round-robin load balancing across all matching pods. For **weighted traffic splitting** (e.g., 90/10 canary), use:

- [Kubernetes Gateway API](https://gateway-api.sigs.k8s.io/) with `HTTPRoute` traffic weights
- [Istio](https://istio.io/) `VirtualService` with weighted destinations
- Scale replica counts (e.g., 9 v1 replicas + 1 v2 replica for ~90/10 split)

These are beyond the scope of this demo.

## Horizontal Pod Autoscaler (HPA)

The [Horizontal Pod Autoscaler](https://kubernetes.io/docs/tasks/run-application/horizontal-pod-autoscale/) scales the number of pod replicas based on observed CPU utilization (or custom metrics). On GKE Autopilot, this creates a two-level autoscaling system:

1. **HPA** scales pods up/down based on CPU metrics
2. **Autopilot** scales nodes up/down to fit the pods

Together they handle full autoscaling — you never manage nodes directly.

```python
autoscaling_v1 = k8s_client.AutoscalingV1Api()

hpa = {
    'apiVersion': 'autoscaling/v1',
    'kind': 'HorizontalPodAutoscaler',
    'metadata': {'name': 'sentiment-hpa', 'namespace': NAMESPACE},
    'spec': {
        'scaleTargetRef': {
            'apiVersion': 'apps/v1',
            'kind': 'Deployment',
            'name': 'sentiment-v2',
        },
        'minReplicas': 1,
        'maxReplicas': 5,
        'targetCPUUtilizationPercentage': 70,
    },
}

try:
    autoscaling_v1.create_namespaced_horizontal_pod_autoscaler(namespace=NAMESPACE, body=hpa)
    print('HPA created')
except Exception as e:
    if '409' in str(e):
        print('HPA already exists')
    else:
        raise
```

```python
# Check HPA status
hpa_status = autoscaling_v1.read_namespaced_horizontal_pod_autoscaler('sentiment-hpa', NAMESPACE)
print(f'HPA: {hpa_status.metadata.name}')
print(f'  Target: {hpa_status.spec.scale_target_ref.name}')
print(f'  Min replicas: {hpa_status.spec.min_replicas}')
print(f'  Max replicas: {hpa_status.spec.max_replicas}')
print(f'  Target CPU: {hpa_status.spec.target_cpu_utilization_percentage}%')
print(f'  Current replicas: {hpa_status.status.current_replicas}')
```

**Autopilot + HPA:** On GKE Autopilot, you do not manage node pools or node scaling. Autopilot provisions and removes nodes automatically based on pod resource requests. The HPA scales pods, and Autopilot scales the underlying infrastructure. This is comparable to Cloud Run's fully managed scaling, but with the full Kubernetes API available for fine-grained control.

## Platform Comparison

| Aspect | GKE Autopilot | Cloud Run | Vertex AI Endpoint |
|--------|---------------|-----------|-------------------|
| Container | Same | Same | Same + Model Registry |
| Cluster management | None (Autopilot) | None | None |
| Scale-to-zero | No (min 1 pod) | Yes | No |
| GPU | Yes (any type) | L4 only | Yes (many types) |
| Networking | Full control (VPC, Ingress, Gateway) | Managed | Managed |
| Traffic splitting | Gateway API / Service selectors | Revisions | Traffic split % |
| Cost model | Per-pod resources | Per-request | Per-replica-hour |
| Kubernetes expertise | Required | None | None |
| Multi-model per node | Yes (pods share nodes) | No | Yes (resource pools) |
| Best for | Teams on K8s, custom networking, GPU | Simple serverless | Managed ML platform |

### When to Choose GKE

**Choose GKE Autopilot when:**
- Your team already operates on Kubernetes and wants ML workloads in the same cluster
- You need custom networking (VPC-native pods, internal load balancers, Gateway API with weighted routing)
- You need GPU scheduling with fine-grained control (specific GPU types, multi-GPU, fractional GPU)
- You want multi-model pods sharing node resources for cost efficiency
- Compliance requirements mandate Kubernetes-level RBAC, network policies, or workload identity

**Choose Cloud Run when:**
- You want the simplest possible deployment with scale-to-zero
- You do not need Kubernetes features and prefer a pure serverless model
- Your workload is bursty with long idle periods (pay-per-request billing)

**Choose Vertex AI Endpoints when:**
- You need ML platform features: model registry, versioning, monitoring, explainability, batch prediction
- You want managed traffic splitting with percentage-based controls
- You prefer a fully managed ML serving solution with no infrastructure decisions
