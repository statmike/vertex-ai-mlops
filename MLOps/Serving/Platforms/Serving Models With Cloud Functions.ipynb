# Serving Models With Cloud Functions

Part of [vertex-ai-mlops/MLOps/Serving](https://github.com/statmike/vertex-ai-mlops/tree/main/MLOps/Serving)

## Overview

Deploy a sentiment analysis model as a **Cloud Function (2nd gen)** — the lightest serverless option. No container image, no Dockerfile, no build step — just a Python function deployed from source.

**The serverless spectrum:**
- **Cloud Functions** (this notebook) — lightest: deploy a Python function directly
- **Cloud Run** — middle: deploy a Docker container with full control
- **Vertex AI Endpoints** — heaviest: managed ML serving with monitoring, model registry, and traffic splitting

**What you'll learn:**
- Cloud Functions structure: `main.py` entry point + `requirements.txt`
- Model loading in global scope — loaded once on cold start, reused for warm invocations
- Cold start optimization strategies (min instances, smaller models)
- Concurrency and memory constraints
- Size constraints (~1-2 GB practical limit due to memory)

**Key concept:** Cloud Functions 2nd gen is backed by Cloud Run under the hood, but with a simpler deployment model. You trade container flexibility for deployment simplicity.

**Runtime:** ~5-10 minutes (deploy + cold start testing).

## Download Model to GCS

Cloud Functions have tighter memory constraints than Cloud Run or Vertex AI (max 32 GB, but practical limit ~2 GB for cold start download time). We use **DistilBERT** — a smaller, distilled version of BERT — which fits comfortably within these constraints.

- **DistilBERT:** ~268 MB — fast to download on cold start
- **BERT:** ~440 MB — would work but slower cold starts

The model is downloaded from HuggingFace and uploaded to GCS. The Cloud Function will download it from GCS on first invocation.

````python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

gcs_prefix = GCS_MODEL_PATH

# Check if already uploaded
existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
if existing:
    print(f'Model already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
else:
    print('Downloading DistilBERT (distilbert-base-uncased-finetuned-sst-2-english) ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(
            'distilbert-base-uncased-finetuned-sst-2-english'
        )
        tokenizer = AutoTokenizer.from_pretrained(
            'distilbert-base-uncased-finetuned-sst-2-english'
        )
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        # Upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')
````

````python
blobs = list(bucket.list_blobs(prefix=GCS_MODEL_PATH))
print(f'Model files at gs://{GCS_BUCKET}/{GCS_MODEL_PATH}: {len(blobs)} files')
for blob in blobs:
    print(f'  {blob.name} ({blob.size:,} bytes)')
````

## Function Source Code

A Cloud Function needs two files:
- **`main.py`** — the Python code with an entry point function decorated with `@functions_framework.http`
- **`requirements.txt`** — Python dependencies (installed automatically during deployment)

**Model loading pattern:** The model is loaded in the **global scope** (outside the request handler). This means:
- **Cold start:** Model downloads from GCS and loads into memory (~30-60s)
- **Warm invocation:** Model is already in memory, only inference runs (~100-500ms)

This is the standard pattern for Cloud Functions with ML models — load once, predict many times.

````python
# Create source directory
source_dir = f'{DIR}/source'
os.makedirs(source_dir, exist_ok=True)
print(f'Source directory: {source_dir}')
````

````python
%%writefile {DIR}/source/main.py
import json
import os
import functions_framework
from google.cloud import storage
import tempfile

# Global scope — loaded once on cold start, reused for warm invocations
model = None
tokenizer = None

def load_model():
    """Load model from GCS on first invocation."""
    global model, tokenizer
    if model is not None:
        return

    from transformers import AutoModelForSequenceClassification, AutoTokenizer
    import torch

    # Download model from GCS
    bucket_name = os.environ['MODEL_BUCKET']
    model_prefix = os.environ['MODEL_PREFIX']

    client = storage.Client()
    bucket = client.bucket(bucket_name)

    tmpdir = tempfile.mkdtemp()
    blobs = bucket.list_blobs(prefix=model_prefix)
    for blob in blobs:
        local_path = os.path.join(tmpdir, os.path.relpath(blob.name, model_prefix))
        os.makedirs(os.path.dirname(local_path), exist_ok=True)
        blob.download_to_filename(local_path)

    model = AutoModelForSequenceClassification.from_pretrained(tmpdir)
    tokenizer = AutoTokenizer.from_pretrained(tmpdir)
    model.eval()
    print(f'Model loaded from gs://{bucket_name}/{model_prefix}')

# Load model eagerly at startup so that min-instances keeps it warm.
# With lazy loading (only inside predict()), min-instances keeps the
# container alive but the model isn't loaded until the first request —
# giving the same ~30s latency as a cold start.
if os.environ.get('MODEL_BUCKET'):
    load_model()

@functions_framework.http
def predict(request):
    """HTTP Cloud Function for sentiment prediction."""
    load_model()

    import torch

    request_json = request.get_json(silent=True)
    if not request_json or 'instances' not in request_json:
        return json.dumps({'error': 'Request must include "instances" field'}), 400

    instances = request_json['instances']
    texts = [inst.get('text', '') for inst in instances]

    inputs = tokenizer(texts, return_tensors='pt', padding=True, truncation=True, max_length=128)

    with torch.no_grad():
        outputs = model(**inputs)

    probs = torch.nn.functional.softmax(outputs.logits, dim=1)
    id2label = model.config.id2label

    predictions = []
    for prob in probs:
        idx = prob.argmax().item()
        predictions.append({
            'label': id2label[idx],
            'score': round(prob[idx].item(), 4),
        })

    return json.dumps({'predictions': predictions})
````

````python
%%writefile {DIR}/source/requirements.txt
functions-framework==3.*
google-cloud-storage
transformers
torch
sentencepiece
````

**Why global scope?** Cloud Functions reuse the same instance for multiple requests. By loading the model outside the request handler, the ~30-60 second download+load cost is paid only on the first request (cold start). Subsequent requests find the model already in memory and return in milliseconds.

This is equivalent to the `prestart.sh` pattern in Cloud Run containers — the difference is that Cloud Functions gives you a simpler mechanism (just put it in module scope or a lazy-init function).

## Deploy Cloud Function

Cloud Functions 2nd gen (`--gen2`) runs on Cloud Run infrastructure under the hood, which gives you:
- Higher resource limits (up to 32 GB memory, 8 vCPUs)
- Concurrency support (multiple requests per instance)
- Longer timeouts (up to 60 minutes)
- Cloud Run networking features

Deployment is source-based — Cloud Build compiles the source into a container image automatically. You never see or manage the container.

````python
!gcloud functions deploy {FUNCTION_NAME} \
    --gen2 \
    --runtime=python311 \
    --region={REGION} \
    --source={DIR}/source \
    --entry-point=predict \
    --trigger-http \
    --allow-unauthenticated \
    --memory=2Gi \
    --timeout=300s \
    --set-env-vars=MODEL_BUCKET={GCS_BUCKET},MODEL_PREFIX={GCS_MODEL_PATH} \
    --min-instances=0 \
    --max-instances=5 \
    --concurrency=10 \
    --project={PROJECT_ID} \
    --quiet
````

Deployment takes ~2-3 minutes. Cloud Build creates a container image from your source code, pushes it to Artifact Registry, and deploys it as a Cloud Run service — all automatically.

````python
function = functions_client.get_function(
    name=f'projects/{PROJECT_ID}/locations/{REGION}/functions/{FUNCTION_NAME}',
)
FUNCTION_URL = function.service_config.uri
print(f'Function URL: {FUNCTION_URL}')
print(f'State: {function.state.name}')
print(f'Runtime: {function.build_config.runtime}')
print(f'Memory: {function.service_config.available_memory}')
````

## Test Predictions

Test the deployed function with HTTP requests. The first request triggers a cold start (model download from GCS + model loading), while subsequent requests reuse the loaded model.

````python
import requests

payload = {
    'instances': [
        {'text': 'This movie is absolutely wonderful!'},
        {'text': 'The food was terrible and the service was even worse.'},
        {'text': 'An average experience, nothing special.'},
    ]
}

# Cold start request — includes model download + load
start = time.time()
response = requests.post(
    FUNCTION_URL,
    json=payload,
)
cold_latency = time.time() - start

print(f'Cold start latency: {cold_latency:.2f}s')
print(f'Status: {response.status_code}')
print(json.dumps(response.json(), indent=2))
````

````python
# Warm request — model already loaded in memory
start = time.time()
response = requests.post(
    FUNCTION_URL,
    json=payload,
)
warm_latency = time.time() - start

print(f'Warm invocation latency: {warm_latency:.2f}s')
print(f'Speedup: {cold_latency / warm_latency:.1f}x faster than cold start')
print(json.dumps(response.json(), indent=2))
````

Notice the dramatic difference:
- **Cold start:** ~30-60 seconds (download model from GCS, load into memory, run inference)
- **Warm invocation:** ~0.1-0.5 seconds (model already in memory, just inference)

This is the fundamental trade-off of scale-to-zero serverless: you save cost when idle, but pay with latency on the first request after scaling down.

````python
# Test with more diverse inputs
diverse_payload = {
    'instances': [
        {'text': 'I absolutely love this product, best purchase ever!'},
        {'text': 'Complete waste of money, do not buy.'},
        {'text': 'It works as expected, nothing more nothing less.'},
        {'text': 'The customer support team was incredibly helpful and friendly.'},
        {'text': 'Broken on arrival, very disappointed.'},
    ]
}

start = time.time()
response = requests.post(FUNCTION_URL, json=diverse_payload)
latency = time.time() - start

print(f'Batch prediction latency: {latency:.2f}s ({len(diverse_payload["instances"])} instances)')
print()
for inst, pred in zip(diverse_payload['instances'], response.json()['predictions']):
    print(f'  "{inst["text"][:55]}..."')
    print(f'    => {pred["label"]} (score: {pred["score"]})')
    print()
````

## Cold Start Optimization

Cold starts are the primary challenge with serverless ML serving. Here are strategies to reduce cold start latency, ordered by impact:

### 1. Min Instances (Strongest)
Set `--min-instances=1` to keep one instance warm at all times. Eliminates cold starts entirely but incurs cost even when idle.

### 2. Smaller Model
DistilBERT (~268 MB) vs BERT (~440 MB) — download time from GCS is a significant portion of cold start latency. Smaller models = faster cold starts.

### 3. ONNX Runtime
Convert PyTorch models to ONNX format for faster loading. ONNX Runtime has lower startup overhead than PyTorch.

### 4. GCS Download Optimization
The GCS download is often the bottleneck. Strategies:
- Use a single archive file (`.tar.gz`) instead of many small files
- Store model files in the same region as the function
- Use GCS composite objects for parallel download

````python
# Redeploy with min-instances=1 to eliminate cold starts
!gcloud functions deploy {FUNCTION_NAME} \
    --gen2 \
    --runtime=python311 \
    --region={REGION} \
    --source={DIR}/source \
    --entry-point=predict \
    --trigger-http \
    --allow-unauthenticated \
    --memory=2Gi \
    --timeout=300s \
    --set-env-vars=MODEL_BUCKET={GCS_BUCKET},MODEL_PREFIX={GCS_MODEL_PATH} \
    --min-instances=1 \
    --max-instances=5 \
    --concurrency=10 \
    --project={PROJECT_ID} \
    --quiet
````

## Comparison: The Serverless Spectrum

### Cloud Functions vs Cloud Run vs Vertex AI Endpoints

| Aspect | Cloud Functions | Cloud Run | Vertex AI Endpoint |
|--------|----------------|-----------|-------------------|
| **Deployment unit** | Python function | Docker container | Container + Model Registry |
| **Build step** | None (source deploy) | Dockerfile + Cloud Build | Dockerfile + Cloud Build |
| **Scale-to-zero** | Yes | Yes | No |
| **Min instances** | Optional | Optional | Required (min 1) |
| **Max model size** | ~1-2 GB (memory-bound) | Multi-GB | Multi-GB |
| **GPU** | No | Yes (L4) | Yes (many types) |
| **Traffic splitting** | No | Yes (revisions) | Yes (traffic split) |
| **Concurrency** | Per-instance | Per-instance | Per-replica |
| **Auth** | IAM / allow-unauth | IAM / allow-unauth | IAM / API key |
| **Monitoring** | Cloud Logging | Cloud Logging | Vertex AI + Cloud Monitoring |
| **Best for** | Small models, webhooks, prototypes | Production containers | Production ML with monitoring |

### Decision Flowchart

```
Need to serve an ML model?
|
+-- Model > 2 GB or needs GPU?
|   |-- Yes --> Cloud Run or Vertex AI
|   |-- No  --> Continue
|
+-- Need model registry, monitoring, or explainability?
|   |-- Yes --> Vertex AI Endpoints
|   |-- No  --> Continue
|
+-- Need traffic splitting or canary deployments?
|   |-- Yes --> Cloud Run (revision-based) or Vertex AI (model-based)
|   |-- No  --> Continue
|
+-- Want simplest possible deployment?
|   |-- Yes --> Cloud Functions (this notebook)
|   |-- No  --> Cloud Run (full container control)
```

**Cloud Functions** is ideal for:
- **Prototyping** — fastest path from model to endpoint
- **Small models** — DistilBERT, small classifiers, embeddings
- **Webhooks** — Slack bots, GitHub webhooks, event-driven inference
- **Low-traffic endpoints** — scale-to-zero saves cost
- **Teams without container expertise** — no Docker knowledge required
