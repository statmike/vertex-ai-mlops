# Spanner ML Remote Model on Vertex AI Endpoint

Deploy a HuggingFace sentiment analysis model to a Vertex AI Endpoint, then call it from Cloud Spanner SQL using `ML.PREDICT()`. This notebook demonstrates:

- Building and deploying a custom FastAPI prediction container to Vertex AI
- Creating a Spanner instance (Enterprise edition) and GoogleSQL database
- Registering the Vertex AI endpoint as a remote model in Spanner
- Running SQL-based ML predictions with `ML.PREDICT()`

**Spanner ML** brings globally distributed ML predictions directly into your database. Predictions run where your data lives -- no ETL, no separate scoring pipeline, no round-trip latency.

**CRITICAL: This notebook uses GoogleSQL dialect (NOT PostgreSQL).** Spanner's PostgreSQL dialect has significant limitations for ML features -- it cannot dynamically build JSON payloads from table data, making `spanner.ml_predict_row()` impractical for real workloads. GoogleSQL's `ML.PREDICT()` works seamlessly with any table query.

**Requirements:**
- **Enterprise edition** -- Spanner ML features require Enterprise or Enterprise Plus edition
- **No IAM permissions needed** -- unlike AlloyDB, Spanner handles auth to Vertex AI internally; no service agent IAM grants required

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params)
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params)

## Download HuggingFace Models

Download two sentiment analysis models from HuggingFace and upload them to GCS. Both models are fine-tuned on the [SST-2](https://huggingface.co/datasets/stanfordnlp/sst2) dataset for binary sentiment classification (POSITIVE/NEGATIVE).

- **v1 (DistilBERT):** Smaller, faster — good for initial deployment
- **v2 (BERT):** Larger, more accurate — uploaded to GCS for further experimentation

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        # upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    # store the GCS URI
    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

````python
for version, info in MODELS.items():
    blobs = list(bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/models/{version}'))
    print(f'{version} ({info["display_name"]}): {len(blobs)} files at {info["artifact_uri"]}')
````

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)

for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']

    # Handle both flat strings (direct calls) and structured objects
    # (Spanner ML sends {"text": "value"} per instance based on INPUT schema)
    texts = [
        inst if isinstance(inst, str) else inst['text']
        for inst in instances
    ]

    results = classifier(texts)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
````

## Create Spanner Instance and Database

[Cloud Spanner](https://cloud.google.com/spanner) is Google Cloud's fully managed, globally distributed, strongly consistent relational database. It uniquely combines:

- **Global distribution** with strong consistency (no eventual consistency)
- **Horizontal scalability** to petabyte scale
- **99.999% availability SLA** for multi-region configurations
- **SQL support** with ACID transactions
- **Built-in Vertex AI integration** for ML predictions from SQL

**Enterprise edition** is required for ML features (`ML.PREDICT()`, `CREATE MODEL`). Standard edition does not support Spanner ML.

This notebook creates a minimal regional instance (1 node) to keep costs low. In production, Spanner scales horizontally by adding nodes.

````python
from google.cloud import spanner_admin_instance_v1 as spanner_admin

# The high-level spanner_v1.Client.instance() doesn't support the edition
# parameter, so we use the admin client directly for Enterprise instances.
admin_client = spanner_admin.InstanceAdminClient()

instance_pb = spanner_admin.Instance(
    name=f'projects/{PROJECT_ID}/instances/{SPANNER_INSTANCE}',
    config=f'projects/{PROJECT_ID}/instanceConfigs/{SPANNER_CONFIG}',
    display_name=SPANNER_INSTANCE,
    node_count=1,
    edition=spanner_admin.Instance.Edition.ENTERPRISE,
)

operation = admin_client.create_instance(
    parent=f'projects/{PROJECT_ID}',
    instance_id=SPANNER_INSTANCE,
    instance=instance_pb,
)
print('Creating Spanner instance (Enterprise edition) ...')
operation.result()
print(f'Instance created: {SPANNER_INSTANCE}')

# Get a high-level instance reference for database operations
instance = spanner_client.instance(SPANNER_INSTANCE)
````

````python
database = instance.database(SPANNER_DATABASE)
operation = database.create()
operation.result()
print(f'Database created: {SPANNER_DATABASE}')
````

## Register Remote Model in Spanner

Spanner ML lets you register a Vertex AI endpoint as a remote model using `CREATE MODEL ... REMOTE OPTIONS(endpoint = '...')`. Once registered, you can call the model from SQL with `ML.PREDICT()` -- just like BigQuery ML.

The `INPUT` and `OUTPUT` schemas must match your endpoint's request/response format exactly. For our sentiment analysis model:
- **INPUT**: a single `text` column (STRING)
- **OUTPUT**: `label` (STRING) and `score` (FLOAT64)

**CRITICAL: GoogleSQL dialect required.** Spanner supports two SQL dialects (GoogleSQL and PostgreSQL), but ML features work reliably only with GoogleSQL. The PostgreSQL dialect's `spanner.ml_predict_row()` function cannot dynamically build JSON payloads from table data, making it impractical for real workloads. This notebook uses GoogleSQL exclusively.

````python
endpoint_resource = endpoint.resource_name
ENDPOINT_URI = f'//aiplatform.googleapis.com/{endpoint_resource}'

# Create the remote model using DDL
model_ddl = f"""
CREATE MODEL SentimentModel
INPUT (text STRING(MAX))
OUTPUT (label STRING(MAX), score FLOAT64)
REMOTE OPTIONS (
    endpoint = '{ENDPOINT_URI}'
)
"""

operation = database.update_ddl([model_ddl])
operation.result()
print('Remote model registered in Spanner')
````

The `INPUT`/`OUTPUT` schema tells Spanner how to serialize requests and deserialize responses. The column names and types must match exactly:
- `text STRING(MAX)` — Spanner sends each instance as `{"text": "value"}` to the endpoint
- `label STRING(MAX)` and `score FLOAT64` match the `predictions` response fields

If your endpoint has a different schema, adjust the `CREATE MODEL` DDL accordingly.

## Create Test Data

Create a table in Spanner and insert sample review sentences for sentiment analysis.

````python
# Create table
table_ddl = """
CREATE TABLE Reviews (
    id INT64 NOT NULL,
    text STRING(MAX),
) PRIMARY KEY (id)
"""
operation = database.update_ddl([table_ddl])
operation.result()

# Insert test data
with database.batch() as batch:
    batch.insert(
        table='Reviews',
        columns=['id', 'text'],
        values=[
            (1, 'This movie is absolutely wonderful and I loved every minute of it!'),
            (2, 'The food was terrible and the service was even worse.'),
            (3, 'An average experience, nothing special but not bad either.'),
            (4, 'Outstanding performance by the entire cast!'),
            (5, 'I would not recommend this to anyone.'),
        ],
    )
print('Test data inserted')
````

## Run Predictions with ML.PREDICT()

Spanner's `ML.PREDICT()` function works just like BigQuery ML's -- pass a model and a subquery (or table reference), and get predictions back as additional columns. This is the core of Spanner ML: SQL-native inference with no application code needed.

### Single Prediction

````python
with database.snapshot() as snapshot:
    results = snapshot.execute_sql("""
        SELECT text, label, score
        FROM ML.PREDICT(
            MODEL SentimentModel,
            (SELECT 'This movie is absolutely wonderful!' AS text)
        )
    """)
    for row in results:
        print(f'Text: {row[0]}')
        print(f'Label: {row[1]}, Score: {row[2]}')
````

### Score All Rows from Reviews Table

````python
with database.snapshot() as snapshot:
    results = snapshot.execute_sql("""
        SELECT r.id, r.text, p.label, p.score
        FROM ML.PREDICT(
            MODEL SentimentModel,
            TABLE Reviews
        ) AS p
        JOIN Reviews AS r ON r.id = p.id
    """)
    for row in results:
        print(f'[{row[0]}] {row[1][:50]}... => {row[2]} ({row[3]:.4f})')
````

### Write Predictions to a Results Table

````python
# Create a results table
results_ddl = """
CREATE TABLE ReviewPredictions (
    id INT64 NOT NULL,
    text STRING(MAX),
    label STRING(MAX),
    score FLOAT64,
) PRIMARY KEY (id)
"""
operation = database.update_ddl([results_ddl])
operation.result()

# Read predictions and write to results table
with database.snapshot() as snapshot:
    results = snapshot.execute_sql("""
        SELECT r.id, r.text, p.label, p.score
        FROM ML.PREDICT(
            MODEL SentimentModel,
            TABLE Reviews
        ) AS p
        JOIN Reviews AS r ON r.id = p.id
    """)
    rows = list(results)

with database.batch() as batch:
    batch.insert(
        table='ReviewPredictions',
        columns=['id', 'text', 'label', 'score'],
        values=rows,
    )

print(f'Wrote {len(rows)} predictions to ReviewPredictions table')

# Verify
with database.snapshot() as snapshot:
    results = snapshot.execute_sql('SELECT id, label, score FROM ReviewPredictions ORDER BY id')
    for row in results:
        print(f'  [{row[0]}] {row[1]} ({row[2]:.4f})')
````

### Globally Distributed Inference

One of Spanner's unique advantages is global distribution. In a multi-region configuration, your data is replicated across regions and `ML.PREDICT()` works from any replica. This means:

- **Low-latency predictions** from any region -- data and inference are co-located
- **No cross-region round trips** -- the Spanner node calls the nearest Vertex AI endpoint
- **Automatic failover** -- if one region goes down, predictions continue from another

For global applications (e-commerce, gaming, financial services), this eliminates the need to build separate scoring pipelines per region.
