# AlloyDB AI Remote Model on Vertex AI Endpoint

Part of [vertex-ai-mlops/MLOps/Serving/Remote Models](https://github.com/statmike/vertex-ai-mlops/tree/main/MLOps/Serving/SQL%20Inference)

## Overview

Deploy a HuggingFace sentiment analysis model to a **Vertex AI Endpoint** and call it directly from **AlloyDB for PostgreSQL** using the built-in `google_ml_integration` extension. This pattern brings ML predictions into your database without any application code — just SQL.

## What You'll Learn

- Build a custom FastAPI serving container and deploy it to a Vertex AI Endpoint
- Create an AlloyDB cluster and primary instance with private IP connectivity
- Enable the `google_ml_integration` extension in AlloyDB
- Register a Vertex AI Endpoint as a model in AlloyDB with `google_ml.create_model()`
- Run row-level predictions from SQL with `google_ml.predict_row()`
- Score a table of rows using SQL-only inference

## Key Concept: AlloyDB Is OLTP

AlloyDB is an **Online Transaction Processing** (OLTP) database. Predictions happen at the *row level*, inside transactions, at insert or update time. This is fundamentally different from BigQuery's batch-oriented `ML.PREDICT()`:

| | AlloyDB | BigQuery |
|---|---|---|
| **Workload** | OLTP — row-level, low-latency | OLAP — batch analytics |
| **Prediction call** | `google_ml.predict_row()` | `ML.PREDICT()` |
| **Typical use** | Score a row at INSERT/UPDATE time | Score millions of rows in a warehouse |
| **Latency** | Milliseconds per row | Seconds to minutes per query |

Both call the same Vertex AI Endpoint — they just serve different operational patterns.

## Download HuggingFace Models

Download two sentiment analysis models from HuggingFace and upload them to GCS. Both models are fine-tuned on the [SST-2](https://huggingface.co/datasets/stanfordnlp/sst2) dataset for binary sentiment classification (POSITIVE/NEGATIVE).

- **v1 (DistilBERT):** Smaller, faster — good for initial deployment
- **v2 (BERT):** Larger, more accurate — the model we'll swap to

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        # upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    # store the GCS URI
    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
````

````python
for version, info in MODELS.items():
    blobs = list(bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/models/{version}'))
    print(f'{version} ({info["display_name"]}): {len(blobs)} files at {info["artifact_uri"]}')
````

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)

for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)

@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}

@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']
    results = classifier(instances)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
````

## AlloyDB Cluster Setup

[AlloyDB for PostgreSQL](https://cloud.google.com/alloydb/docs/overview) is Google Cloud's fully managed, PostgreSQL-compatible database. It delivers up to 4x the performance of standard PostgreSQL for transactional workloads.

**Private IP requirement:** AlloyDB instances require a private IP in your VPC network. This section sets up:
1. **Private Services Access (PSA)** — reserves an IP range for Google-managed services
2. **VPC peering** — connects your VPC to Google's service network
3. **AlloyDB cluster** — the regional container for instances
4. **Primary instance** — the compute resource running PostgreSQL

> **Important:** AlloyDB charges continuously while running (~$0.30/hour for the minimal config below). The cleanup section at the end deletes all resources.

### Private Services Access (VPC Networking)

````python
# Reserve an IP range for Google-managed services (idempotent)
!gcloud compute addresses describe google-managed-services-{NETWORK} \
    --global --project={PROJECT_ID} 2>/dev/null || \
gcloud compute addresses create google-managed-services-{NETWORK} \
    --global --purpose=VPC_PEERING --prefix-length=16 \
    --network={NETWORK} --project={PROJECT_ID}

# Create or update the private connection
# (update is needed if a peering already exists with a different range)
!gcloud services vpc-peerings connect \
    --service=servicenetworking.googleapis.com \
    --ranges=google-managed-services-{NETWORK} \
    --network={NETWORK} --project={PROJECT_ID} 2>/dev/null || \
gcloud services vpc-peerings update \
    --service=servicenetworking.googleapis.com \
    --ranges=google-managed-services-{NETWORK} \
    --network={NETWORK} --project={PROJECT_ID} --force 2>/dev/null || true

print('Private Services Access configured')
````

### Create AlloyDB Cluster

````python
parent = f'projects/{PROJECT_ID}/locations/{REGION}'
cluster_name = f'{parent}/clusters/{ALLOYDB_CLUSTER}'

try:
    cluster = alloydb_client.get_cluster(name=cluster_name)
    print(f'Cluster already exists: {cluster.name}')
except Exception:
    cluster_config = types.Cluster(
        network=f'projects/{PROJECT_ID}/global/networks/{NETWORK}',
        initial_user=types.UserPassword(
            user=DB_USER,
            password=ALLOYDB_PASSWORD,
        ),
    )
    operation = alloydb_client.create_cluster(
        parent=parent,
        cluster_id=ALLOYDB_CLUSTER,
        cluster=cluster_config,
    )
    print('Creating AlloyDB cluster (this takes ~10 minutes)...')
    cluster = operation.result()
    print(f'Created: {cluster.name}')
````

### Create Primary Instance

````python
instance_name = f'{cluster.name}/instances/{ALLOYDB_INSTANCE}'

try:
    instance = alloydb_client.get_instance(name=instance_name)
    print(f'Instance already exists: {instance.name}')
except Exception:
    instance_config = types.Instance(
        instance_type=types.Instance.InstanceType.PRIMARY,
        machine_config=types.Instance.MachineConfig(cpu_count=2),
        availability_type=types.Instance.AvailabilityType.ZONAL,
    )
    operation = alloydb_client.create_instance(
        parent=cluster.name,
        instance_id=ALLOYDB_INSTANCE,
        instance=instance_config,
    )
    print('Creating primary instance (this takes ~10 minutes)...')
    instance = operation.result()
    print(f'Created: {instance.name}')
````

## Enable google_ml_integration

The [`google_ml_integration`](https://cloud.google.com/alloydb/docs/ai/invoke-predictions) extension is a built-in AlloyDB feature that lets you call Vertex AI endpoints (and other Google AI services) directly from SQL. Key functions:

- `google_ml.create_model()` — register an external endpoint as a named model
- `google_ml.predict_row()` — call the model for a single row and return JSON
- `google_ml.drop_model()` — unregister a model

This section enables the extension, grants IAM permissions, and registers the Vertex AI endpoint.

### Batch Predictions Over a Table

````python
# Create a test table with sample text
with engine.connect() as conn:
    conn.execute(sqlalchemy.text('''
        DROP TABLE IF EXISTS reviews;
        CREATE TABLE reviews (
            id SERIAL PRIMARY KEY,
            text TEXT NOT NULL
        );
        INSERT INTO reviews (text) VALUES
            ('This movie is absolutely wonderful and I loved every minute of it!'),
            ('The food was terrible and the service was even worse.'),
            ('An average experience, nothing special but not bad either.'),
            ('Best purchase I have ever made. Highly recommend!'),
            ('Do not waste your money on this product.'),
            ('The hotel was clean and the staff was friendly.'),
            ('I was very disappointed with the quality.'),
            ('A solid choice for the price. Good value overall.');
    '''))
    conn.commit()
    print('Created reviews table with 8 sample rows')
````

````python
# Score every row in the reviews table
score_sql = """
SELECT
    id,
    text,
    google_ml.predict_row(
        'sentiment_model',
        json_build_object('instances', json_build_array(text))
    ) AS prediction
FROM reviews
ORDER BY id;
"""

with engine.connect() as conn:
    result = conn.execute(sqlalchemy.text(score_sql))
    for row in result:
        pred = row[2] if isinstance(row[2], dict) else json.loads(row[2])
        label = pred['predictions'][0]['label']
        score = pred['predictions'][0]['score']
        print(f'  [{row[0]}] {label:>8s} ({score:.4f})  {row[1][:60]}')
````

### Real-Time Pattern: Prediction at INSERT Time

In a production OLTP system you would typically score rows automatically using a PostgreSQL trigger. The trigger fires on each INSERT and stores the prediction alongside the data:

```sql
-- Add a sentiment column
ALTER TABLE reviews ADD COLUMN sentiment TEXT;
ALTER TABLE reviews ADD COLUMN sentiment_score FLOAT;

-- Create a trigger function
CREATE OR REPLACE FUNCTION score_review()
RETURNS TRIGGER AS $$
DECLARE
    pred JSON;
BEGIN
    pred := google_ml.predict_row(
        'sentiment_model',
        json_build_object('instances', json_build_array(NEW.text))
    );
    NEW.sentiment       := pred->'predictions'->0->>'label';
    NEW.sentiment_score := (pred->'predictions'->0->>'score')::FLOAT;
    RETURN NEW;
END;
$$ LANGUAGE plpgsql;

-- Attach the trigger
CREATE TRIGGER trg_score_review
    BEFORE INSERT ON reviews
    FOR EACH ROW EXECUTE FUNCTION score_review();

-- Now every INSERT gets scored automatically
INSERT INTO reviews (text) VALUES ('This is amazing!');
SELECT * FROM reviews WHERE id = (SELECT MAX(id) FROM reviews);
```

This keeps latency low (single Vertex AI call per row) and ensures every row has a prediction the moment it enters the database.

### AlloyDB vs BigQuery ML — Choosing the Right Pattern

| | AlloyDB (`google_ml.predict_row`) | BigQuery (`ML.PREDICT`) |
|---|---|---|
| **Workload** | OLTP — low-latency, row-level | OLAP — batch analytics |
| **Typical scale** | Hundreds/thousands of rows per second | Millions/billions of rows per job |
| **Trigger support** | Yes — score at INSERT/UPDATE time | No — runs as a query |
| **Best for** | Operational apps, real-time scoring | Dashboards, periodic batch scoring |
| **Cost model** | Per-second compute (instance running) | Per-query bytes scanned |

Both call the **same Vertex AI Endpoint** — the difference is whether you need transactional, row-level predictions (AlloyDB) or warehouse-scale batch scoring (BigQuery).
