# BQML Import Model via ONNX

This notebook demonstrates the *other* way to use ML models in BigQuery: **importing the model itself** into BigQuery rather than calling an external endpoint. The model runs natively inside BigQuery -- no endpoint needed.

## Overview

We take a HuggingFace PyTorch sentiment analysis model (DistilBERT), convert it to ONNX format, import the ONNX file into BigQuery, and run predictions with `ML.PREDICT()`.

**Key difference from Remote Models:** with a Remote Model the model runs on an external Vertex AI endpoint and BigQuery calls out to it. With an ONNX import the model runs *inside* BigQuery compute -- there is no endpoint to deploy, manage, or pay for.

**Tradeoffs:**

| ONNX Import (this notebook) | Remote Model |
|------------------------------|--------------|
| Model must be < 250 MB | No size limit |
| Inputs must be ARRAY (pre-tokenized) | Endpoint handles any format |
| Caller needs the tokenizer for NLP models | Tokenization inside serving container |
| No endpoint to manage or pay for | Endpoint always running (cost) |
| BQ compute only | BQ compute + endpoint cost |
| Best for numeric/tabular inputs | Best for NLP and complex serving |

## What You'll Learn

1. Download a HuggingFace PyTorch model (DistilBERT)
2. Export it to ONNX format with `torch.onnx.export()`
3. Validate the ONNX model locally with ONNX Runtime
4. Upload the ONNX file to Cloud Storage
5. Import it into BigQuery with `CREATE MODEL ... OPTIONS(model_type='ONNX')`
6. Run predictions with `ML.PREDICT()`

## Download HuggingFace Model

We use **DistilBERT** (`distilbert-base-uncased-finetuned-sst-2-english`) for this notebook because it fits comfortably within BigQuery's **250 MB limit** for imported ONNX models. The larger BERT variant (`textattack/bert-base-uncased-SST-2`) used in the Remote Models notebooks would be too large after ONNX export.

```python
MODEL_NAME = 'distilbert-base-uncased-finetuned-sst-2-english'

model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model.eval()
print(f'Model loaded: {MODEL_NAME}')
print(f'Labels: {model.config.id2label}')
```

---
## Convert to ONNX

[ONNX](https://onnx.ai/) (Open Neural Network Exchange) is a framework-agnostic model format. By exporting our PyTorch model to ONNX we can import it directly into BigQuery ML, which only accepts ONNX (not PyTorch `.pt` files).

The key steps:
1. Create a dummy tokenized input so PyTorch can trace the computation graph
2. Call `torch.onnx.export()` with dynamic batch axes so the model can handle variable batch sizes
3. Verify the exported ONNX model produces the same predictions

```python
# Create dummy input
dummy_text = "This is a test sentence."
inputs = tokenizer(dummy_text, return_tensors='pt', padding='max_length', max_length=128, truncation=True)

# Export to ONNX
ONNX_PATH = f'{DIR}/model.onnx'

torch.onnx.export(
    model,
    (inputs['input_ids'], inputs['attention_mask']),
    ONNX_PATH,
    input_names=['input_ids', 'attention_mask'],
    output_names=['logits'],
    dynamic_axes={
        'input_ids': {0: 'batch_size'},
        'attention_mask': {0: 'batch_size'},
        'logits': {0: 'batch_size'},
    },
    opset_version=17,
)

# Post-process for BigQuery compatibility:
# 1. Convert weights to float16 — DistilBERT's float32 weights are ~256 MB,
#    which exceeds BigQuery's 250 MB ONNX import limit.
# 2. Downgrade IR version to 8 — BigQuery's ONNX runtime (v1.12) only
#    supports IR version <= 8; the dynamo exporter produces IR 10.
from onnxruntime.transformers import float16

onnx_model = onnx.load(ONNX_PATH)
onnx_model = float16.convert_float_to_float16(onnx_model, keep_io_types=True)
onnx_model.ir_version = 8
onnx.save(onnx_model, ONNX_PATH)

size_mb = os.path.getsize(ONNX_PATH) / (1024 * 1024)
print(f'ONNX model saved: {ONNX_PATH} ({size_mb:.1f} MB)')
print(f'IR version: {onnx_model.ir_version}, opset: {onnx_model.opset_import[0].version}')
assert size_mb < 250, f'Model too large for BQML import: {size_mb:.1f} MB (limit: 250 MB)'
```

```python
import onnxruntime as ort

session = ort.InferenceSession(ONNX_PATH)

# Run inference
ort_inputs = {
    'input_ids': inputs['input_ids'].numpy(),
    'attention_mask': inputs['attention_mask'].numpy(),
}
logits = session.run(None, ort_inputs)[0]
predicted_class = np.argmax(logits, axis=1)[0]
print(f'Test prediction: {model.config.id2label[predicted_class]}')
print(f'Logits: {logits[0]}')
```

> **Note on dynamic axes:** We set `dynamic_axes` for the batch dimension so that `ML.PREDICT()` can score multiple rows at once. Without this the ONNX model would only accept a single-row input.

## Import ONNX Model into BigQuery

`CREATE MODEL ... OPTIONS(model_type='ONNX')` tells BigQuery to load the ONNX file from GCS and register it as a native BigQuery ML model. Once imported the model runs entirely inside BigQuery -- no external endpoint.

```python
# Create dataset
dataset = bigquery.Dataset(f'{PROJECT_ID}.{BQ_DATASET}')
dataset.location = BQ_LOCATION
bq.create_dataset(dataset, exists_ok=True)

# Import ONNX model
import_sql = f"""
CREATE OR REPLACE MODEL `{BQ_DATASET}.sentiment_onnx`
OPTIONS (
    model_type = 'ONNX',
    model_path = 'gs://{GCS_BUCKET}/{GCS_ONNX_PATH}'
)
"""
bq.query(import_sql).result()
print('ONNX model imported into BigQuery')
```

```python
# ML.FEATURE_INFO is not supported for ONNX models, so we use the
# Python client to verify the model was imported successfully.
model_ref = bq.get_model(f'{PROJECT_ID}.{BQ_DATASET}.sentiment_onnx')
print(f'Model: {model_ref.model_id}')
print(f'Type: {model_ref.model_type}')
print(f'Created: {model_ref.created}')
```

## Run Predictions with ML.PREDICT()

**Important:** ONNX models imported into BigQuery expect **pre-tokenized ARRAY inputs**, not raw text. For a text model like DistilBERT this means you must tokenize externally (using the *same* HuggingFace tokenizer the model was trained with) and pass the resulting integer arrays into the query. BigQuery has no built-in tokenizer that matches HuggingFace's vocabulary.

### Why not import the tokenizer too?

A natural question: could we import the tokenizer as a *separate* BigQuery model and chain the two with nested `ML.PREDICT()` calls?

- **As an ONNX model?** No -- ONNX's standard ops don't support the string processing and vocabulary lookup needed for WordPiece/BPE tokenization. Projects like `onnxruntime-extensions` add custom tokenizer ops, but BigQuery's ONNX runtime only supports standard ops.
- **As a TensorFlow SavedModel?** Maybe -- TensorFlow has much better string support (`tf.strings.*`, `tf.lookup.StaticHashTable`), and TF Hub offers pre-built BERT preprocessing SavedModels. In theory you could import a TF tokenizer and chain it: `ML.PREDICT(onnx_model, (SELECT ... FROM ML.PREDICT(tf_tokenizer, ...)))`. In practice, BigQuery's sandboxed TF Serving environment may not support the ops these models need (hash table initialization, `tensorflow_text` custom ops). Worth experimenting with, but not reliable today.
- **As a BigQuery Remote Function?** This works -- deploy a lightweight Cloud Function that only tokenizes (no GPU needed), register it as a Remote Function, and call it in your query. But you're managing external infrastructure again, which is what ONNX import was supposed to avoid.

**Bottom line:** ONNX import is most practical for models with **numeric inputs** (tabular, time-series, embeddings) where the data is already in the right format. For NLP models that need tokenization, a **Remote Model** with tokenization handled inside the serving container is usually the better choice.

```python
def tokenize_for_bq(texts, max_length=128):
    """Tokenize texts and format for BigQuery ONNX model input."""
    encoded = tokenizer(
        texts,
        padding='max_length',
        max_length=max_length,
        truncation=True,
        return_tensors='np',
    )
    rows = []
    for i in range(len(texts)):
        rows.append({
            'input_ids': encoded['input_ids'][i].tolist(),
            'attention_mask': encoded['attention_mask'][i].tolist(),
        })
    return rows
```

```python
test_text = 'This movie is absolutely wonderful!'
tokens = tokenize_for_bq([test_text])[0]

predict_sql = f"""
SELECT logits
FROM ML.PREDICT(
    MODEL `{BQ_DATASET}.sentiment_onnx`,
    (SELECT {tokens['input_ids']} AS input_ids, {tokens['attention_mask']} AS attention_mask)
)
"""
result = bq.query(predict_sql).result().to_dataframe()
result
```

```python
# Batch prediction: create a temp table with pre-tokenized data
batch_texts = [
    'This movie is absolutely wonderful!',
    'The food was terrible and the service was even worse.',
    'An average experience, nothing special but not bad either.',
    'I love spending time outdoors on a sunny day.',
    'That was the worst decision I have ever made.',
]
batch_tokens = tokenize_for_bq(batch_texts)

# Build UNION ALL query with one row per text
row_sqls = []
for i, tok in enumerate(batch_tokens):
    row_sqls.append(
        f"SELECT {i} AS row_id, {tok['input_ids']} AS input_ids, {tok['attention_mask']} AS attention_mask"
    )
input_query = ' UNION ALL\n'.join(row_sqls)

batch_sql = f"""
SELECT *
FROM ML.PREDICT(
    MODEL `{BQ_DATASET}.sentiment_onnx`,
    ({input_query})
)
ORDER BY row_id
"""
batch_result = bq.query(batch_sql).result().to_dataframe()

# Show results alongside original text
for i, text in enumerate(batch_texts):
    row = batch_result[batch_result['row_id'] == i].iloc[0]
    print(f'Text: "{text}"')
    print(f'  logits: {row["logits"]}')
    print()
```

```python
# The ONNX model outputs logits, need softmax + argmax for labels
postprocess_sql = f"""
WITH predictions AS (
    SELECT
        input_ids,
        logits
    FROM ML.PREDICT(
        MODEL `{BQ_DATASET}.sentiment_onnx`,
        (SELECT {tokens['input_ids']} AS input_ids, {tokens['attention_mask']} AS attention_mask)
    )
)
SELECT
    logits,
    CASE
        WHEN logits[OFFSET(0)] > logits[OFFSET(1)] THEN 'NEGATIVE'
        ELSE 'POSITIVE'
    END AS predicted_label
FROM predictions
"""
bq.query(postprocess_sql).result().to_dataframe()
```

### ONNX Import vs Remote Model

| Aspect | ONNX Import | Remote Model |
|--------|-------------|--------------|
| Model runs in | BigQuery compute | External endpoint |
| Model size limit | < 250 MB | Unlimited |
| Endpoint needed | No | Yes (always running) |
| Input format | ARRAY (pre-tokenized) | Any (endpoint handles) |
| Tokenization | External -- caller needs the tokenizer | Inside the serving container |
| Supported frameworks | ONNX-exportable only | Any |
| Pre/post processing | SQL or pre-tokenize | In serving container |
| Cost | BQ compute only | BQ compute + endpoint |
| Latency | BQ query time | BQ query + endpoint call |
| Best for | Small models with numeric inputs | Large models, NLP, complex serving |
