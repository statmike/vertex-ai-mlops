# BQML Remote Model on Vertex AI Endpoint

Part of `vertex-ai-mlops/MLOps/Serving/Remote Models`

## Overview

This notebook shows how to deploy a sentiment analysis model to a **Vertex AI Endpoint** and then call it directly from SQL using **BigQuery ML Remote Models** with `ML.PREDICT()`. The result is a fully SQL-accessible inference pipeline — no Python required at scoring time.

## What You Will Learn

- **Cloud Resource Connection** — create and configure the BigQuery connection that authenticates to Vertex AI
- **CREATE MODEL REMOTE** — register a Vertex AI Endpoint as a BQML remote model with explicit INPUT / OUTPUT schema
- **ML.PREDICT()** — run single-row and multi-row predictions from pure SQL
- **Batch scoring patterns** — score entire tables, apply business logic, and persist results

## Prerequisites

- A Google Cloud project with billing enabled
- The following APIs enabled: BigQuery, BigQuery Connection, Vertex AI, Artifact Registry, Cloud Build
- Familiarity with BigQuery SQL basics

## What Are BQML Remote Models?

BigQuery ML remote models let BigQuery call **external prediction endpoints** during query execution. BigQuery sends rows to the endpoint via a Cloud Resource Connection and merges the returned predictions back into the query result set.

Key properties:
- Works with **any model framework** served behind a Vertex AI Endpoint (PyTorch, TensorFlow, scikit-learn, XGBoost, custom containers, etc.)
- Works with **any model size** — the model runs on the endpoint infrastructure, not inside BigQuery
- Supports `ML.PREDICT()` for inference and scheduled queries for automated batch scoring
- The connection's service account handles authentication — no keys or tokens in SQL

```python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
```

```python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'

    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        # upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    # store the GCS URI
    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}'
```

## Build and Deploy Container

This section builds a lightweight **FastAPI** custom container for the sentiment model. The same container pattern is used throughout the MLOps/Serving/Online series.

The container:
1. Downloads the model from GCS at startup
2. Exposes `/health` and `/predict` routes expected by Vertex AI
3. Accepts `{"instances": [{"text": "..."}]}` and returns `{"predictions": [{"label": "...", "score": 0.99}]}`


```python
%%writefile {DIR}/source/Dockerfile
FROM python:3.11-slim

COPY ./app /app
COPY ./requirements.txt requirements.txt

RUN pip install --no-cache-dir --upgrade pip \
  && pip install --no-cache-dir -r requirements.txt

ENV AIP_HEALTH_ROUTE=/health
ENV AIP_PREDICT_ROUTE=/predict
ENV AIP_HTTP_PORT=8080

CMD ["uvicorn", "app.main:app", "--host", "0.0.0.0", "--port", "8080"]
```

```python
%%writefile {DIR}/source/app/main.py
import os, json, tempfile
from fastapi import FastAPI, Request
from google.cloud import storage
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch

app = FastAPI()

# ── Load model at startup ──────────────────────────────────────────
MODEL_DIR = tempfile.mkdtemp()
ARTIFACT_URI = os.environ.get('AIP_STORAGE_URI', '')

if ARTIFACT_URI:
    # Download from GCS
    client = storage.Client()
    prefix = ARTIFACT_URI.replace('gs://', '').split('/', 1)
    bucket = client.bucket(prefix[0])
    blobs = bucket.list_blobs(prefix=prefix[1] if len(prefix) > 1 else '')
    for blob in blobs:
        dest = os.path.join(
            MODEL_DIR,
            os.path.relpath(blob.name, prefix[1] if len(prefix) > 1 else ''),
        )
        os.makedirs(os.path.dirname(dest), exist_ok=True)
        blob.download_to_filename(dest)

model = AutoModelForSequenceClassification.from_pretrained(MODEL_DIR)
tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR)
model.eval()
LABELS = list(model.config.id2label.values())


# ── Routes ──────────────────────────────────────────────────────────
@app.get(os.environ.get('AIP_HEALTH_ROUTE', '/health'), status_code=200)
def health():
    return {}


@app.post(os.environ.get('AIP_PREDICT_ROUTE', '/predict'))
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']

    texts = [inst['text'] if isinstance(inst, dict) else inst for inst in instances]

    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        logits = model(**inputs).logits
    probs = torch.softmax(logits, dim=-1)
    scores, indices = probs.max(dim=-1)

    predictions = []
    for score, idx in zip(scores, indices):
        predictions.append({
            'label': LABELS[idx.item()],
            'score': round(score.item(), 6),
        })
    return {'predictions': predictions}
```

## BigQuery Cloud Resource Connection

A **Cloud Resource Connection** is how BigQuery authenticates to external services like Vertex AI.

When you create a connection, BigQuery provisions a dedicated Google-managed service account. This service account is what actually calls the Vertex AI Endpoint on behalf of your SQL queries. You grant it the minimum permissions needed — in this case, `roles/aiplatform.user` so it can invoke predictions.

Key points:
- The connection is region-scoped (we use `US` to match BigQuery dataset location)
- The service account is automatically managed — no keys to rotate
- A single connection can be reused across multiple remote models


```python
connection = types.Connection(
    cloud_resource=types.CloudResourceProperties(),
)

parent = f'projects/{PROJECT_ID}/locations/{BQ_LOCATION}'
connection_id = f'{SERIES}_{EXPERIMENT.replace("-", "_")}'

try:
    conn = conn_client.get_connection(name=f'{parent}/connections/{connection_id}')
    print(f'Connection already exists: {conn.name}')
except Exception:
    conn = conn_client.create_connection(
        parent=parent,
        connection_id=connection_id,
        connection=connection,
    )
    print(f'Created connection: {conn.name}')

CONNECTION_SA = conn.cloud_resource.service_account_id
print(f'Connection service account: {CONNECTION_SA}')
```

## Register Remote Model in BigQuery

The `CREATE MODEL ... REMOTE WITH CONNECTION` statement registers a Vertex AI Endpoint as a BQML model. After this step, any BigQuery user with access to the model can run predictions with plain SQL.

The `INPUT` and `OUTPUT` clauses define the schema contract between BigQuery and the endpoint:
- **INPUT** — the columns BigQuery will send as the `instances` payload
- **OUTPUT** — the fields BigQuery expects in each element of the `predictions` response


```python
# Create dataset if needed
dataset = bigquery.Dataset(f'{PROJECT_ID}.{BQ_DATASET}')
dataset.location = BQ_LOCATION
bq.create_dataset(dataset, exists_ok=True)
print(f'Dataset ready: {BQ_DATASET}')

ENDPOINT_RESOURCE = endpoint.resource_name
print(f'Endpoint resource: {ENDPOINT_RESOURCE}')

# Create remote model
create_model_sql = f"""
CREATE OR REPLACE MODEL `{BQ_DATASET}.sentiment_remote`
INPUT (text STRING)
OUTPUT (label STRING, score FLOAT64)
REMOTE WITH CONNECTION `{PROJECT_ID}.{BQ_LOCATION}.{connection_id}`
OPTIONS (
    endpoint = 'https://{REGION}-aiplatform.googleapis.com/v1/{ENDPOINT_RESOURCE}'
)
"""
print(create_model_sql)
bq.query(create_model_sql).result()
print('Remote model created')
```

### INPUT / OUTPUT Schema Mapping

The schema you declare in `CREATE MODEL` must match what the endpoint expects and returns:

| Direction | BQML Declaration | Endpoint JSON |
|-----------|-----------------|---------------|
| INPUT | `text STRING` | `{"instances": [{"text": "..."}]}` |
| OUTPUT | `label STRING, score FLOAT64` | `{"predictions": [{"label": "POSITIVE", "score": 0.99}]}` |

BigQuery automatically wraps each row's input columns into an element of the `instances` array and unpacks each element of `predictions` back into output columns.

## Run Predictions with ML.PREDICT()

`ML.PREDICT()` is the SQL function that sends rows to the remote model and returns predictions. It works like any other table-valued function — you can wrap it in `SELECT`, `JOIN`, `WHERE`, `GROUP BY`, etc.


```python
query = f"""
SELECT *
FROM ML.PREDICT(
    MODEL `{BQ_DATASET}.sentiment_remote`,
    (SELECT 'This movie is absolutely wonderful!' AS text)
)
"""
bq.query(query).result().to_dataframe()
```

```python
query = f"""
SELECT *
FROM ML.PREDICT(
    MODEL `{BQ_DATASET}.sentiment_remote`,
    (
        SELECT text
        FROM UNNEST([
            'This movie is absolutely wonderful!',
            'The food was terrible and the service was even worse.',
            'An average experience, nothing special but not bad either.',
            'I can\\'t believe how great this product is — highly recommend!',
            'Worst purchase I have ever made. Total waste of money.'
        ]) AS text
    )
)
"""
bq.query(query).result().to_dataframe()
```

```python
# Create a test table with review text
create_table_sql = f"""
CREATE OR REPLACE TABLE `{BQ_DATASET}.reviews` AS
SELECT *
FROM UNNEST([
    STRUCT('rev-001' AS review_id, 'electronics' AS category, 'Absolutely love this phone, best purchase ever!' AS text),
    STRUCT('rev-002', 'electronics', 'Screen broke after one week. Very disappointed.'),
    STRUCT('rev-003', 'food', 'The chocolate cake was divine, will order again.'),
    STRUCT('rev-004', 'food', 'Arrived stale and the packaging was damaged.'),
    STRUCT('rev-005', 'travel', 'Beautiful hotel with stunning ocean views.'),
    STRUCT('rev-006', 'travel', 'Flight was delayed 6 hours with no communication.'),
    STRUCT('rev-007', 'electronics', 'Does exactly what it says. Good value for the price.'),
    STRUCT('rev-008', 'food', 'Tastes okay but nothing special for the price.'),
    STRUCT('rev-009', 'travel', 'Friendly staff but the room was a bit small.'),
    STRUCT('rev-010', 'electronics', 'Returned it immediately. Completely unusable.')
])
"""
bq.query(create_table_sql).result()
print(f'Created table {BQ_DATASET}.reviews with 10 rows')
```

```python
# Score the reviews table and add business logic
score_sql = f"""
CREATE OR REPLACE TABLE `{BQ_DATASET}.scored_reviews` AS
SELECT
    r.review_id,
    r.category,
    r.text,
    p.label AS sentiment,
    p.score AS confidence,
    CASE
        WHEN p.label = 'NEGATIVE' AND p.score > 0.95 THEN 'URGENT'
        WHEN p.label = 'NEGATIVE' THEN 'REVIEW'
        ELSE 'OK'
    END AS action
FROM ML.PREDICT(
    MODEL `{BQ_DATASET}.sentiment_remote`,
    (SELECT review_id, category, text FROM `{BQ_DATASET}.reviews`)
) p
JOIN `{BQ_DATASET}.reviews` r USING (review_id, category, text)
ORDER BY p.score DESC
"""
bq.query(score_sql).result()

# Display results
bq.query(f'SELECT * FROM `{BQ_DATASET}.scored_reviews` ORDER BY action, confidence DESC').result().to_dataframe()
```

### Remote Model vs Model Import (ONNX)

| Aspect | Remote Model (this notebook) | BQML Model Import (ONNX) |
|--------|------------------------------|--------------------------|
| **Where model runs** | Vertex AI Endpoint (your infrastructure) | Inside BigQuery (Google-managed) |
| **Model size** | Any size | < 250 MB (ONNX limit) |
| **Framework** | Any (PyTorch, TF, custom) | Must export to ONNX |
| **Latency** | Network round-trip to endpoint | In-process, very fast |
| **Cost** | Endpoint compute + BQ query | BQ query only (on-demand slots) |
| **Best for** | Large models, GPU inference, existing endpoints | Small models, high-throughput batch scoring |
