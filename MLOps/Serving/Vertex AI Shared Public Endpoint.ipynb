# Vertex AI Shared Public Endpoint

Deploy HuggingFace sentiment analysis models to a [Shared Public Endpoint](https://cloud.google.com/vertex-ai/docs/predictions/choose-endpoint-type) on Vertex AI. This notebook demonstrates:

- Building a custom FastAPI prediction container with Cloud Build
- Creating a shared public endpoint (the simplest endpoint type)
- Deploying, swapping, and undeploying models with zero-downtime traffic splitting

**When to use shared?** Shared public endpoints are the simplest to set up and are the only endpoint type that supports tuned Gemini model deployment and AutoML models with explainability. They use a shared networking plane, which means lower limits: 1.5 MB payload, 60-second timeout, 30K QPM quota, HTTP only (no gRPC or streaming). For most production workloads, consider a [dedicated public endpoint](./Vertex%20AI%20Dedicated%20Public%20Endpoint.ipynb) instead.

**Story arc:**
1. Deploy DistilBERT (100% traffic)
2. Deploy BERT alongside (0% traffic)
3. Shift 100% traffic to BERT
4. Undeploy DistilBERT — endpoint keeps serving

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params)
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params)

## Download HuggingFace Models

Download two sentiment analysis models from HuggingFace and upload them to GCS. Both models are fine-tuned on the [SST-2](https://huggingface.co/datasets/stanfordnlp/sst2) dataset for binary sentiment classification (POSITIVE/NEGATIVE).

- **v1 (DistilBERT):** Smaller, faster — good for initial deployment
- **v2 (BERT):** Larger, more accurate — the model we'll swap to

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
    },
}
````

````python
for version, info in MODELS.items():
    gcs_prefix = f'{SERIES}/{EXPERIMENT}/models/{version}'
    
    # check if already uploaded
    existing = list(bucket.list_blobs(prefix=gcs_prefix, max_results=1))
    if existing:
        print(f'{version} ({info["hub_name"]}) already in GCS at gs://{GCS_BUCKET}/{gcs_prefix}')
        continue
    
    print(f'Downloading {version}: {info["hub_name"]} ...')
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)
        
        # upload to GCS
        for root, dirs, files in os.walk(tmpdir):
            for fname in files:
                local_path = os.path.join(root, fname)
                blob_name = f'{gcs_prefix}/{os.path.relpath(local_path, tmpdir)}'
                bucket.blob(blob_name).upload_from_filename(local_path)
        print(f'  Uploaded to gs://{GCS_BUCKET}/{gcs_prefix}')

    # store the GCS URI
    info['artifact_uri'] = f'gs://{GCS_BUCKET}/{gcs_prefix}'

# set artifact_uri for any that were already uploaded
for version, info in MODELS.items():
    if 'artifact_uri' not in info:
        info['artifact_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/models/{version}' 
````

````python
for version, info in MODELS.items():
    blobs = list(bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/models/{version}'))
    print(f'{version} ({info["display_name"]}): {len(blobs)} files at {info["artifact_uri"]}')
````

## Build Custom Prediction Container

Build a custom container with [FastAPI](https://fastapi.tiangolo.com/) that:
1. Downloads the model files from GCS at startup (via `AIP_STORAGE_URI`)
2. Loads a HuggingFace `transformers` sentiment analysis pipeline
3. Serves predictions at the Vertex AI prediction route

```
|__ Dockerfile
|__ requirements.txt
|__ app
    |__ __init__.py
    |__ main.py
    |__ prestart.sh
```

### Setup Artifact Registry

````python
docker_repo = None
for repo in ar.list_repositories(parent=f'projects/{PROJECT_ID}/locations/{REGION}'):
    if AR_REPO == repo.name.split('/')[-1]:
        docker_repo = repo
        print(f'Retrieved existing repo: {docker_repo.name}')

if not docker_repo:
    operation = ar.create_repository(
        request=artifactregistry_v1.CreateRepositoryRequest(
            parent=f'projects/{PROJECT_ID}/locations/{REGION}',
            repository_id=AR_REPO,
            repository=artifactregistry_v1.Repository(
                description=f'Docker images for the {SERIES} series.',
                name=AR_REPO,
                format_=artifactregistry_v1.Repository.Format.DOCKER,
                labels={'series': SERIES},
            ),
        )
    )
    print('Creating Repository ...')
    docker_repo = operation.result()
    print(f'Completed creating repo: {docker_repo.name}')
````

````python
REPOSITORY = f"{REGION}-docker.pkg.dev/{PROJECT_ID}/{docker_repo.name.split('/')[-1]}"
REPOSITORY
````

### Create Application Files

````python
%%writefile {DIR}/source/Dockerfile
FROM tiangolo/uvicorn-gunicorn-fastapi:python3.10

COPY ./app /app
COPY ./requirements.txt requirements.txt

RUN pip install --no-cache-dir --upgrade pip \
  && pip install --no-cache-dir -r requirements.txt
````

````python
%%writefile {DIR}/source/requirements.txt
google-cloud-storage
transformers
torch
sentencepiece
````

````python
%%writefile {DIR}/source/app/__init__.py
# init file
````

````python
%%writefile {DIR}/source/app/main.py
import os
from fastapi import FastAPI, Request
from transformers import pipeline
from google.cloud import storage

app = FastAPI()

# Download model files from GCS (set by AIP_STORAGE_URI)
gcs = storage.Client()
storage_uri = os.environ['AIP_STORAGE_URI']
parts = storage_uri.replace('gs://', '').split('/', 1)
bucket = gcs.bucket(parts[0])
prefix = parts[1] if len(parts) > 1 else ''

model_dir = '/tmp/model'
os.makedirs(model_dir, exist_ok=True)
for blob in bucket.list_blobs(prefix=prefix):
    rel = blob.name[len(prefix):].lstrip('/')
    if rel:
        local = os.path.join(model_dir, rel)
        os.makedirs(os.path.dirname(local), exist_ok=True)
        blob.download_to_filename(local)

# Load the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model_dir, tokenizer=model_dir)


@app.get(os.environ['AIP_HEALTH_ROUTE'], status_code=200)
def health():
    return {}


@app.post(os.environ['AIP_PREDICT_ROUTE'])
async def predict(request: Request):
    body = await request.json()
    instances = body['instances']
    results = classifier(instances)
    return {
        'predictions': [
            {'label': r['label'], 'score': round(r['score'], 6)}
            for r in results
        ]
    }
````

````python
%%writefile {DIR}/source/app/prestart.sh
#!/bin/bash
export PORT=$AIP_HTTP_PORT
````

### Upload Source to GCS

````python
source_files = [
    'Dockerfile',
    'requirements.txt',
    'app/__init__.py',
    'app/main.py',
    'app/prestart.sh',
]
for f in source_files:
    bucket.blob(f'{SERIES}/{EXPERIMENT}/source/{f}').upload_from_filename(f'{DIR}/source/{f}')

print('Source files uploaded to GCS:')
for blob in bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/source'):
    print(f'  gs://{bucket.name}/{blob.name}')
````

### Build Container with Cloud Build

````python
build = cloudbuild_v1.Build(steps=[])

# Step 1: retrieve source from GCS
build.steps.append({
    'name': 'gcr.io/cloud-builders/gsutil',
    'args': ['cp', '-r', f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/source/*', '/workspace'],
})

# Step 2: docker build
build.steps.append({
    'name': 'gcr.io/cloud-builders/docker',
    'args': ['build', '-t', f'{REPOSITORY}/{EXPERIMENT}', '/workspace'],
})

# Push the image
build.images = [f'{REPOSITORY}/{EXPERIMENT}']

print('Starting Cloud Build ...')
operation = cb.create_build(project_id=PROJECT_ID, build=build)
build_response = operation.result()
print(f'Build status: {build_response.status.name}')
print(f'Image: {build_response.artifacts.images[0]}')
````

````python
IMAGE_URI = build_response.artifacts.images[0]
IMAGE_URI
````

## Upload Models to Vertex AI Model Registry

Upload both models using the same container image but different `artifact_uri` values pointing to each model's GCS directory.

````python
vertex_model_v1 = aiplatform.Model.upload(
    display_name=f'{EXPERIMENT}-distilbert',
    serving_container_image_uri=IMAGE_URI,
    artifact_uri=MODELS['v1']['artifact_uri'],
    labels={'series': SERIES, 'experiment': EXPERIMENT, 'version': 'v1'},
)
print(f'v1 model: {vertex_model_v1.resource_name}')
````

````python
vertex_model_v2 = aiplatform.Model.upload(
    display_name=f'{EXPERIMENT}-bert',
    serving_container_image_uri=IMAGE_URI,
    artifact_uri=MODELS['v2']['artifact_uri'],
    labels={'series': SERIES, 'experiment': EXPERIMENT, 'version': 'v2'},
)
print(f'v2 model: {vertex_model_v2.resource_name}')
````

## Create Shared Public Endpoint

Create a shared public endpoint — the default when `dedicated_endpoint_enabled` is not set. This is the simplest endpoint type with no special configuration needed.

````python
endpoint = aiplatform.Endpoint.create(
    display_name=f'{EXPERIMENT}',
    labels={'series': SERIES, 'experiment': EXPERIMENT},
)
print(f'Endpoint: {endpoint.resource_name}')
````

---
## Deploy DistilBERT (v1) — 100% Traffic

Deploy the smaller DistilBERT model as the initial model serving all traffic.

````python
endpoint.deploy(
    model=vertex_model_v1,
    deployed_model_display_name='distilbert-sst2',
    machine_type='n1-standard-4',
    min_replica_count=1,
    max_replica_count=2,
    traffic_percentage=100,
)
print(f'Deployed DistilBERT with 100% traffic')
print(f'Traffic split: {endpoint.traffic_split}')
````

### Test Sentences

````python
TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]

def show_predictions(response):
    for instance, pred in zip(TEST_INSTANCES, response.predictions):
        print(f'  "{instance[:60]}..."')
        print(f'    => {pred["label"]} (score: {pred["score"]})')
        print()
````

### Test: Predictions from DistilBERT

Send prediction requests to the endpoint. All responses should come from DistilBERT.

````python
print(f'Expected model: DistilBERT (v1)\n')
response = endpoint.predict(instances=TEST_INSTANCES)
show_predictions(response)
````

---
## Deploy BERT (v2) Alongside — 0% Traffic

Deploy the larger BERT model alongside DistilBERT, initially receiving no traffic. This lets the model warm up before we shift traffic to it.

````python
endpoint.deploy(
    model=vertex_model_v2,
    deployed_model_display_name='bert-sst2',
    machine_type='n1-standard-4',
    min_replica_count=1,
    max_replica_count=2,
    traffic_percentage=0,
)
print(f'Deployed BERT alongside DistilBERT with 0% traffic')
print(f'Traffic split: {endpoint.traffic_split}')
````

### Test: Predictions Still Served by DistilBERT

Even though BERT is now deployed, all traffic still goes to DistilBERT.

````python
print(f'Expected model: DistilBERT (v1) — BERT is deployed but at 0% traffic\n')
response = endpoint.predict(instances=TEST_INSTANCES)
show_predictions(response)
````

---
## Shift 100% Traffic to BERT (v2)

Now shift all traffic from DistilBERT to BERT. The endpoint URL remains the same — only the model serving predictions changes.

````python
print(f'Current traffic split: {endpoint.traffic_split}')

# Build new traffic split: 100% to BERT (v2), 0% to DistilBERT (v1)
new_traffic_split = {}
for deployed_model in endpoint.list_models():
    if deployed_model.display_name == 'bert-sst2':
        new_traffic_split[deployed_model.id] = 100
    else:
        new_traffic_split[deployed_model.id] = 0

print(f'New traffic split: {new_traffic_split}')
endpoint.update(traffic_split=new_traffic_split)
print(f'Updated traffic split: {endpoint.traffic_split}')
````

### Test: Predictions Now Served by BERT

Same endpoint, but now the responses come from BERT. Notice the scores differ from DistilBERT — it's a different (larger) model.

> **Label difference:** BERT (v2) returns `LABEL_0`/`LABEL_1` instead of `POSITIVE`/`NEGATIVE` because the [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) model config uses numeric label IDs while DistilBERT's config has named labels. The different labels actually help confirm the traffic shift worked. To control label names, set `id2label` in the model's `config.json` (e.g., `{"0": "NEGATIVE", "1": "POSITIVE"}`) before uploading to GCS.

````python
print(f'Expected model: BERT (v2) — traffic shifted from DistilBERT\n')
response = endpoint.predict(instances=TEST_INSTANCES)
show_predictions(response)
````

---
## Undeploy DistilBERT — Endpoint Still Serving

Remove the DistilBERT model entirely from the endpoint. The endpoint continues to serve predictions without interruption.

````python
# Find and undeploy the DistilBERT model (which now has 0% traffic)
for deployed_model in endpoint.list_models():
    if deployed_model.display_name == 'distilbert-sst2':
        print(f'Undeploying {deployed_model.display_name} (id: {deployed_model.id})')
        endpoint.undeploy(deployed_model_id=deployed_model.id)
        break

print(f'\nRemaining deployed models:')
for m in endpoint.list_models():
    print(f'  {m.display_name} — traffic: {endpoint.traffic_split.get(m.id, 0)}%')
````

### Test: Endpoint Still Works After Undeploy

The endpoint remains fully functional after removing a model. No downtime during the lifecycle.

````python
print(f'Expected model: BERT (v2) — DistilBERT has been undeployed\n')
response = endpoint.predict(instances=TEST_INSTANCES)
show_predictions(response)
print('Endpoint remained stable through deploy, traffic shift, and undeploy!')
````
