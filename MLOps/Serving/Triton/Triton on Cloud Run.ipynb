# Triton on Cloud Run

Deploy Triton Inference Server to [Cloud Run](https://cloud.google.com/run) with L4 GPU for serverless model serving. This notebook demonstrates:

- Deploying the same Triton container from the [Vertex AI notebook](./Triton%20on%20Vertex%20AI%20Endpoints.ipynb) to Cloud Run
- GPU deployment with `nvidia-l4` accelerator
- OIDC authentication (Cloud Run uses ID tokens, not access tokens)
- Warm instance latency characteristics
- Concurrency tuning: feeding Triton's internal dynamic batcher

**Cloud Run vs Vertex AI for Triton:**
- Cloud Run: scale-to-zero, OIDC auth, revision-based traffic splitting, pay-per-use
- Vertex AI: managed ML platform, `rawPredict`, Model Registry integration, min 1 replica

**Story arc:**
1. Build Triton container with ensemble pipeline
2. Deploy to Cloud Run with L4 GPU
3. Test with OIDC-authenticated requests
4. Measure warm instance latency
5. Tune concurrency for Triton's batcher

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params)
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params)

## Build Triton Container

Download models, export to ONNX, create the ensemble pipeline, and build the container. Same model repository as the [Vertex AI notebook](./Triton%20on%20Vertex%20AI%20Endpoints.ipynb) with one difference: Cloud Run expects the server on port 8080 (same as Vertex AI, so the Dockerfile is identical).

````python
MODELS = {
    'v1': {'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english', 'triton_name': 'distilbert_onnx'},
    'v2': {'hub_name': 'textattack/bert-base-uncased-SST-2', 'triton_name': 'bert_onnx'},
}

for version, info in MODELS.items():
    local_dir = f'{MODELS_LOCAL}/{version}'
    if not os.path.exists(local_dir) or not any(os.scandir(local_dir)):
        print(f'Downloading {version}: {info["hub_name"]} ...')
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(local_dir)
        tokenizer.save_pretrained(local_dir)
    else:
        print(f'{version}: already downloaded')

    onnx_dir = f'{MODEL_REPO}/{info["triton_name"]}/1'
    onnx_path = f'{onnx_dir}/model.onnx'
    os.makedirs(onnx_dir, exist_ok=True)
    if not os.path.exists(onnx_path):
        print(f'  Exporting to ONNX ...')
        model = AutoModelForSequenceClassification.from_pretrained(local_dir)
        tokenizer = AutoTokenizer.from_pretrained(local_dir)
        model.eval()
        dummy = tokenizer('Test.', return_tensors='pt', padding='max_length', max_length=128, truncation=True)
        torch.onnx.export(model, (dummy['input_ids'], dummy['attention_mask']), onnx_path,
            input_names=['input_ids', 'attention_mask'], output_names=['logits'],
            dynamic_axes={'input_ids': {0: 'b', 1: 's'}, 'attention_mask': {0: 'b', 1: 's'}, 'logits': {0: 'b'}},
            opset_version=17, do_constant_folding=True)
        print(f'  Exported ({os.path.getsize(onnx_path)/1024/1024:.1f} MB)')
````

````python
import shutil

# ONNX configs (GPU — KIND_GPU for L4)
for name in ['distilbert_onnx', 'bert_onnx']:
    config = f'''name: "{name}"
backend: "onnxruntime"
max_batch_size: 8
input [
  {{ name: "input_ids", data_type: TYPE_INT64, dims: [-1] }},
  {{ name: "attention_mask", data_type: TYPE_INT64, dims: [-1] }}
]
output [ {{ name: "logits", data_type: TYPE_FP32, dims: [2] }} ]
instance_group [ {{ count: 1, kind: KIND_GPU, gpus: [0] }} ]
dynamic_batching {{ preferred_batch_size: [2, 4, 8], max_queue_delay_microseconds: 100000 }}'''
    with open(f'{MODEL_REPO}/{name}/config.pbtxt', 'w') as f:
        f.write(config)

# Tokenizer
os.makedirs(f'{MODEL_REPO}/tokenizer/1/hf_tokenizer', exist_ok=True)
with open(f'{MODEL_REPO}/tokenizer/config.pbtxt', 'w') as f:
    f.write('name: "tokenizer"\nbackend: "python"\nmax_batch_size: 8\ninput [ { name: "TEXT", data_type: TYPE_STRING, dims: [1] } ]\noutput [\n  { name: "input_ids", data_type: TYPE_INT64, dims: [-1] },\n  { name: "attention_mask", data_type: TYPE_INT64, dims: [-1] }\n]\ninstance_group [ { count: 1, kind: KIND_CPU } ]')

with open(f'{MODEL_REPO}/tokenizer/1/model.py', 'w') as f:
    f.write('import numpy as np\nimport triton_python_backend_utils as pb_utils\nfrom transformers import AutoTokenizer\n\nclass TritonPythonModel:\n    def initialize(self, args):\n        self.tokenizer = AutoTokenizer.from_pretrained(f"{args[\'model_repository\']}/{args[\'model_version\']}/hf_tokenizer")\n    def execute(self, requests):\n        responses = []\n        for request in requests:\n            texts = [t.decode(\'utf-8\') for t in pb_utils.get_input_tensor_by_name(request, \'TEXT\').as_numpy().flatten()]\n            encoded = self.tokenizer(texts, padding=\'max_length\', truncation=True, max_length=128, return_tensors=\'np\')\n            responses.append(pb_utils.InferenceResponse([\n                pb_utils.Tensor(\'input_ids\', encoded[\'input_ids\'].astype(np.int64)),\n                pb_utils.Tensor(\'attention_mask\', encoded[\'attention_mask\'].astype(np.int64)),\n            ]))\n        return responses\n    def finalize(self): pass\n')

for f in Path(f'{MODELS_LOCAL}/v1').glob('*'):
    if f.name in ('tokenizer.json', 'tokenizer_config.json', 'vocab.txt', 'special_tokens_map.json'):
        shutil.copy2(f, f'{MODEL_REPO}/tokenizer/1/hf_tokenizer/')

# Postprocessor
os.makedirs(f'{MODEL_REPO}/postprocessor/1', exist_ok=True)
with open(f'{MODEL_REPO}/postprocessor/config.pbtxt', 'w') as f:
    f.write('name: "postprocessor"\nbackend: "python"\nmax_batch_size: 8\ninput [ { name: "logits", data_type: TYPE_FP32, dims: [2] } ]\noutput [\n  { name: "LABEL", data_type: TYPE_STRING, dims: [1] },\n  { name: "SCORE", data_type: TYPE_FP32, dims: [1] }\n]\ninstance_group [ { count: 1, kind: KIND_CPU } ]')

with open(f'{MODEL_REPO}/postprocessor/1/model.py', 'w') as f:
    f.write('import numpy as np\nimport triton_python_backend_utils as pb_utils\n\nclass TritonPythonModel:\n    def initialize(self, args): self.id2label = {0: \'NEGATIVE\', 1: \'POSITIVE\'}\n    def execute(self, requests):\n        responses = []\n        for request in requests:\n            logits = pb_utils.get_input_tensor_by_name(request, \'logits\').as_numpy()\n            exp_l = np.exp(logits - np.max(logits, axis=1, keepdims=True))\n            probs = exp_l / exp_l.sum(axis=1, keepdims=True)\n            pred = probs.argmax(axis=1)\n            labels = np.array([[self.id2label[i]] for i in pred], dtype=object)\n            scores = np.array([[probs[j, pred[j]]] for j in range(len(pred))], dtype=np.float32)\n            responses.append(pb_utils.InferenceResponse([pb_utils.Tensor(\'LABEL\', labels), pb_utils.Tensor(\'SCORE\', scores)]))\n        return responses\n    def finalize(self): pass\n')

# Ensemble
os.makedirs(f'{MODEL_REPO}/sentiment_ensemble/1', exist_ok=True)
with open(f'{MODEL_REPO}/sentiment_ensemble/config.pbtxt', 'w') as f:
    f.write('''name: "sentiment_ensemble"
platform: "ensemble"
max_batch_size: 8
input [ { name: "TEXT", data_type: TYPE_STRING, dims: [1] } ]
output [ { name: "LABEL", data_type: TYPE_STRING, dims: [1] }, { name: "SCORE", data_type: TYPE_FP32, dims: [1] } ]
ensemble_scheduling {
  step [
    { model_name: "tokenizer", model_version: -1, input_map { key: "TEXT", value: "TEXT" }, output_map { key: "input_ids", value: "input_ids" }, output_map { key: "attention_mask", value: "attention_mask" } },
    { model_name: "distilbert_onnx", model_version: -1, input_map { key: "input_ids", value: "input_ids" }, input_map { key: "attention_mask", value: "attention_mask" }, output_map { key: "logits", value: "logits" } },
    { model_name: "postprocessor", model_version: -1, input_map { key: "logits", value: "logits" }, output_map { key: "LABEL", value: "LABEL" }, output_map { key: "SCORE", value: "SCORE" } }
  ]
}''')

# Dockerfile
with open(f'{SOURCE_DIR}/Dockerfile', 'w') as f:
    f.write('FROM nvcr.io/nvidia/tritonserver:24.12-py3\n\nRUN pip install --no-cache-dir transformers sentencepiece\n\nCOPY model_repository /models\n\nENTRYPOINT ["tritonserver", "--model-repository=/models", "--http-port=8080", "--allow-grpc=false", "--allow-metrics=false", "--log-verbose=0"]\n')

print('Model repository and Dockerfile created.')
````

## Deploy to Cloud Run with L4 GPU

Deploy the Triton container to Cloud Run with an L4 GPU. Key settings:
- **GPU:** `nvidia-l4` (24 GB VRAM, good for inference)
- **Min instances:** 1 (avoids cold start from pulling the 15 GB image)
- **Concurrency:** 80 (Triton handles batching internally, so high concurrency is fine)
- **Memory:** 16 GiB (model loading + ONNX sessions)
- **CPU:** 4 (supports GPU workload)

````python
SERVICE_NAME = f'{SERIES}-{EXPERIMENT}'

service = run_v2.Service(
    template=run_v2.RevisionTemplate(
        containers=[
            run_v2.Container(
                image=IMAGE_URI,
                resources=run_v2.ResourceRequirements(
                    limits={'cpu': '4', 'memory': '16Gi', 'nvidia.com/gpu': '1'},
                ),
                ports=[run_v2.ContainerPort(container_port=8080)],
                startup_probe=run_v2.Probe(
                    http_get=run_v2.HTTPGetAction(path='/v2/health/ready', port=8080),
                    initial_delay_seconds=10,
                    period_seconds=10,
                    failure_threshold=30,  # 5 min for model loading
                ),
                liveness_probe=run_v2.Probe(
                    http_get=run_v2.HTTPGetAction(path='/v2/health/live', port=8080),
                    period_seconds=30,
                ),
            ),
        ],
        scaling=run_v2.RevisionScaling(min_instance_count=1, max_instance_count=3),
        max_instance_request_concurrency=80,
        timeout={"seconds": 300},
        node_selector=run_v2.NodeSelector(accelerator='nvidia-l4'),
    ),
    labels={'series': SERIES, 'experiment': EXPERIMENT},
)

print(f'Creating Cloud Run service: {SERVICE_NAME} ...')
operation = run_client.create_service(
    request=run_v2.CreateServiceRequest(
        parent=f'projects/{PROJECT_ID}/locations/{REGION}',
        service=service,
        service_id=SERVICE_NAME,
    )
)
service_response = operation.result()
print(f'Service created: {service_response.uri}')
````

## Authentication: OIDC ID Tokens

Cloud Run uses [OIDC ID tokens](https://cloud.google.com/run/docs/authenticating/service-to-service) for authentication — not the OAuth2 access tokens used by Vertex AI. The ID token's `audience` must match the service URL.

````python
SERVICE_URL = service_response.uri

TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]

# Get OIDC ID token (Cloud Run uses ID tokens, not access tokens)
auth_req = google.auth.transport.requests.Request()
token = google_id_token.fetch_id_token(auth_req, SERVICE_URL)

# Send Triton V2 request
triton_request = {
    "inputs": [{"name": "TEXT", "shape": [len(TEST_INSTANCES), 1], "datatype": "BYTES",
                "data": [[t] for t in TEST_INSTANCES]}],
    "outputs": [{"name": "LABEL"}, {"name": "SCORE"}],
}

response = httpx.post(
    f'{SERVICE_URL}/v2/models/sentiment_ensemble/infer',
    json=triton_request,
    headers={'Authorization': f'Bearer {token}', 'Content-Type': 'application/json'},
    timeout=120,
)

result = response.json()
outputs = {o['name']: o['data'] for o in result['outputs']}

print(f'Cloud Run + L4 GPU (status: {response.status_code}):\n')
for text, label, score in zip(TEST_INSTANCES, outputs['LABEL'], outputs['SCORE']):
    if isinstance(label, list): label = label[0]
    if isinstance(score, list): score = score[0]
    print(f'  "{text[:60]}..."')
    print(f'    => {label} (score: {score:.6f})')
    print()
````

## Warm Instance Latency

With `min_instance_count=1`, the first instance stays warm and avoids cold starts. Cold starts (pulling the ~15 GB image + model initialization) only occur when scaling up additional instances.

The measurements below are for a warm instance — the baseline latency you can expect in production.

> **Tip:** For latency-sensitive workloads, always set `min_instance_count >= 1` to avoid the 15 GB container pull on cold start.

````python
# Measure warm latency vs first-request-after-scale
def measure_latency(n=10):
    times = []
    for _ in range(n):
        token = google_id_token.fetch_id_token(auth_req, SERVICE_URL)
        triton_req = {
            "inputs": [{"name": "TEXT", "shape": [1, 1], "datatype": "BYTES",
                        "data": [["Quick test"]]}],
            "outputs": [{"name": "LABEL"}, {"name": "SCORE"}],
        }
        start = time.time()
        resp = httpx.post(f'{SERVICE_URL}/v2/models/sentiment_ensemble/infer',
                         json=triton_req,
                         headers={'Authorization': f'Bearer {token}', 'Content-Type': 'application/json'},
                         timeout=120)
        times.append((time.time() - start) * 1000)
    return times

latencies = measure_latency()
latencies.sort()
print(f'Warm instance latency (single request):')
print(f'  Median: {latencies[len(latencies)//2]:.0f} ms')
print(f'  P95:    {latencies[int(len(latencies)*0.95)]:.0f} ms')
print(f'  Min:    {latencies[0]:.0f} ms')
print(f'  Max:    {latencies[-1]:.0f} ms')
````

## Concurrency and Triton's Dynamic Batcher

Cloud Run's `max_instance_request_concurrency` controls how many concurrent requests each instance handles. With Triton, set this **high** (e.g., 80) because:

1. Triton has its own internal request queue and dynamic batcher
2. Multiple concurrent requests feed the batcher, which forms batches for efficient GPU inference
3. Low concurrency (e.g., 1) would create one instance per request — wasteful and prevents batching

```
max_instance_request_concurrency = 80
                │
                ▼
    ┌──── Cloud Run Instance ────┐
    │   ┌──── Triton ────────┐   │
    │   │  Dynamic Batcher   │   │
    │   │  ┌──────────────┐  │   │
    │   │  │ Queue (80)   │──┼───┼── batch → GPU inference
    │   │  └──────────────┘  │   │
    │   └────────────────────┘   │
    └────────────────────────────┘
```

````python
import concurrent.futures

def send_one(text):
    token = google_id_token.fetch_id_token(auth_req, SERVICE_URL)
    req = {"inputs": [{"name": "TEXT", "shape": [1, 1], "datatype": "BYTES", "data": [[text]]}],
           "outputs": [{"name": "LABEL"}, {"name": "SCORE"}]}
    start = time.time()
    resp = httpx.post(f'{SERVICE_URL}/v2/models/sentiment_ensemble/infer', json=req,
                     headers={'Authorization': f'Bearer {token}', 'Content-Type': 'application/json'}, timeout=120)
    return (time.time() - start) * 1000, resp.status_code

# Send 16 concurrent requests
texts = ["Great product!"] * 16
print(f'Sending {len(texts)} concurrent requests ...')
start = time.time()
with concurrent.futures.ThreadPoolExecutor(max_workers=16) as executor:
    futures = [executor.submit(send_one, t) for t in texts]
    results = [f.result() for f in futures]
total = (time.time() - start) * 1000

latencies = [r[0] for r in results]
statuses = [r[1] for r in results]
latencies.sort()

print(f'All {len(texts)} completed in {total:.0f} ms')
print(f'Individual latencies: median={latencies[len(latencies)//2]:.0f} ms, max={latencies[-1]:.0f} ms')
print(f'Throughput: {len(texts) / (total/1000):.1f} requests/sec')
print(f'All status 200: {all(s == 200 for s in statuses)}')
````

## CPU-Only Alternative

The same container works without a GPU — just remove the `nvidia.com/gpu` resource limit and the `node_selector`. Triton automatically uses CPU for ONNX inference (our configs have `KIND_GPU` but Triton falls back gracefully, or you can set `KIND_CPU`).

CPU deployment is significantly cheaper and supports scale-to-zero, but inference is slower.

> **To deploy CPU-only:** Create a new revision without GPU resources. This is left as an exercise — the pattern is identical to the [Cloud Run Platforms notebook](../Platforms/Serving%20Models%20on%20Cloud%20Run.ipynb).

## Cloud Run vs GKE for Triton

| Aspect | Cloud Run | GKE Autopilot |
|--------|:---:|:---:|
| **Scale-to-zero** | Yes | Yes (with config) |
| **GPU support** | L4 | Any GPU type |
| **Cold start** | High (~15 GB pull) | Lower (pre-pulled images) |
| **Autoscaling** | Concurrency-based | HPA (custom metrics) |
| **Triton metrics** | Not exposed | Prometheus scraping |
| **Model repo updates** | Rebuild container | GCS FUSE hot-reload |
| **Networking** | Managed | Full K8s control |
| **Cost model** | Pay-per-request | Pay-per-pod |
| **Ops burden** | Low | Medium |
| **Best for** | Bursty/intermittent traffic | Sustained high-throughput |
