# Triton Inference Server — Fundamentals

Run [NVIDIA Triton Inference Server](https://developer.nvidia.com/triton-inference-server) locally with Docker. This notebook introduces Triton's core concepts — model repository, `config.pbtxt`, and multi-backend serving — using two approaches:

- **Python backend:** Wraps a HuggingFace `transformers.pipeline` — accepts raw text, returns labels and scores. The simplest way to deploy any Python model on Triton.
- **ONNX backend:** Exports the same model to ONNX format for high-performance inference. Takes pre-tokenized inputs (motivating the ensemble pattern in the next notebook).

**What you'll learn:**
1. Build a Triton model repository with Python and ONNX backends

6. Understand model versioning and instance groups

## What is Triton Inference Server?

[NVIDIA Triton Inference Server](https://developer.nvidia.com/triton-inference-server) is an open-source serving platform that standardizes ML model deployment. Instead of writing custom serving code (FastAPI, Flask, etc.), you place model files into a **model repository** directory and Triton handles the rest — loading, versioning, batching, and serving via HTTP and gRPC.

**Key differences from FastAPI-based serving:**

| Aspect | Custom FastAPI | Triton Inference Server |
|--------|---------------|------------------------|
| **Serving code** | You write it (routes, I/O parsing) | None — Triton handles it |
| **Model loading** | Your code loads the model | Triton loads from model repository |
| **Batching** | You implement it | Built-in dynamic batching |
| **Multi-model** | Multiple containers or routes | One server, many models |
| **Versioning** | Manual (container tags, GCS paths) | Built-in version directories |
| **Protocol** | Custom REST endpoints | Standard KServe V2 protocol (HTTP + gRPC) |
| **Multi-framework** | One framework per container | ONNX, TensorFlow, PyTorch, Python, TensorRT in one server |

**Model Repository Structure:**

```
model_repository/
├── model_name_a/
│   ├── config.pbtxt          ← model configuration
│   ├── 1/                    ← version 1
│   │   └── model.onnx        ← model artifact
│   └── 2/                    ← version 2
│       └── model.onnx
└── model_name_b/
    ├── config.pbtxt
    └── 1/
        └── model.py           ← Python backend
```

Each model gets a directory with a `config.pbtxt` (defining inputs, outputs, backend, and runtime settings) and numbered version subdirectories containing the model artifacts.

## Python Backend

The [Python backend](https://github.com/triton-inference-server/python_backend) lets you wrap any Python code as a Triton model. You write a `model.py` implementing `initialize()` and `execute()`, and Triton handles the rest.

This is the easiest way to deploy a HuggingFace model on Triton — the `model.py` wraps `transformers.pipeline` just like our FastAPI `main.py` does, but without any HTTP routing code.

### config.pbtxt

The configuration defines:
- **backend:** `"python"` for the Python backend
- **max_batch_size:** Triton can batch up to this many requests together
- **inputs/outputs:** Named tensors with data types and dimensions
- **instance_group:** How many model instances to run, and on what hardware

````python
os.makedirs(f'{MODEL_REPO}/distilbert_python/1', exist_ok=True)

config_text = '''name: "distilbert_python"
backend: "python"
max_batch_size: 8

input [
  {
    name: "TEXT"
    data_type: TYPE_STRING
    dims: [1]
  }
]
output [
  {
    name: "LABEL"
    data_type: TYPE_STRING
    dims: [1]
  },
  {
    name: "SCORE"
    data_type: TYPE_FP32
    dims: [1]
  }
]

instance_group [
  {
    count: 1
    kind: KIND_CPU
  }
]'''

with open(f'{MODEL_REPO}/distilbert_python/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/distilbert_python/config.pbtxt')
````

### model.py

The Python backend model implements two methods:
- **`initialize()`** — Called once when Triton loads the model. Loads the HuggingFace pipeline.
- **`execute()`** — Called for each batch of inference requests. Extracts text from input tensors, runs the pipeline, and returns label/score tensors.

````python
model_py = '''import numpy as np
import triton_python_backend_utils as pb_utils
from transformers import pipeline


class TritonPythonModel:
    def initialize(self, args):
        model_dir = f"{args['model_repository']}/{args['model_version']}/hf_model"
        self.pipeline = pipeline(
            'sentiment-analysis', model=model_dir, tokenizer=model_dir,
            device=-1,  # CPU
        )

    def execute(self, requests):
        responses = []
        for request in requests:
            # Extract text strings from input tensor
            input_tensor = pb_utils.get_input_tensor_by_name(request, 'TEXT')
            texts = [t.decode('utf-8') for t in input_tensor.as_numpy().flatten()]

            # Run sentiment analysis
            results = self.pipeline(texts)

            # Build output tensors
            labels = np.array([[r['label']] for r in results], dtype=object)
            scores = np.array([[r['score']] for r in results], dtype=np.float32)

            out_label = pb_utils.Tensor('LABEL', labels)
            out_score = pb_utils.Tensor('SCORE', scores)
            responses.append(pb_utils.InferenceResponse([out_label, out_score]))
        return responses

    def finalize(self):
        pass
'''

with open(f'{MODEL_REPO}/distilbert_python/1/model.py', 'w') as f:
    f.write(model_py)
print(f'Wrote {MODEL_REPO}/distilbert_python/1/model.py')
````

## ONNX Backend

The [ONNX Runtime backend](https://github.com/triton-inference-server/onnxruntime_backend) runs models in ONNX format — a framework-neutral representation that enables hardware-optimized inference. ONNX models are typically faster than Python backend models because they bypass the Python GIL and run in optimized C++ runtimes.

**Trade-off:** ONNX models take pre-tokenized inputs (integer tensors), not raw text. The client must handle tokenization. This motivates the ensemble pattern in the next notebook, where Triton chains a Python tokenizer → ONNX model → Python postprocessor server-side.

### Export to ONNX

Use `torch.onnx.export()` to convert both HuggingFace models to ONNX format with dynamic batch and sequence dimensions.

> **Dynamic shapes:** PyTorch's default ONNX exporter (dynamo-based) traces the model with the dummy input and specializes shapes it observes. If the dummy input has `batch_size=1`, the exporter bakes that into internal operations, making the model batch=1 only. The fix: use a **multi-sample dummy input** (batch>1) so the exporter keeps the batch dimension dynamic.

````python
from torch.export import Dim

batch = Dim('batch', min=1, max=8)
seq = Dim('seq', min=1, max=512)

# Multi-sample dummy input — prevents the exporter from specializing batch to 1
dummy_texts = ['Sentence one.', 'Sentence two.', 'Sentence three.', 'Sentence four.']

for name, version in [('distilbert_onnx', 'v1'), ('bert_onnx', 'v2')]:
    model_dir = f'{MODELS_LOCAL}/{version}'
    onnx_dir = f'{MODEL_REPO}/{name}/1'
    onnx_path = f'{onnx_dir}/model.onnx'
    os.makedirs(onnx_dir, exist_ok=True)

    if os.path.exists(onnx_path):
        print(f'{name}: ONNX model already exists at {onnx_path}')
        continue

    print(f'Exporting {name} from {model_dir} ...')
    hf_model = AutoModelForSequenceClassification.from_pretrained(model_dir)
    hf_tokenizer = AutoTokenizer.from_pretrained(model_dir)
    hf_model.eval()

    dummy = hf_tokenizer(
        dummy_texts, return_tensors='pt',
        padding='max_length', max_length=128, truncation=True,
    )

    torch.onnx.export(
        hf_model,
        (dummy['input_ids'], dummy['attention_mask']),
        onnx_path,
        input_names=['input_ids', 'attention_mask'],
        output_names=['logits'],
        dynamic_shapes=({0: batch, 1: seq}, {0: batch, 1: seq}),
    )

    # Report total size (graph + optional external weight data)
    total_bytes = sum(
        os.path.getsize(os.path.join(onnx_dir, f))
        for f in os.listdir(onnx_dir)
    )
    print(f'  Exported to {onnx_path} ({total_bytes / 1024 / 1024:.1f} MB)')

    # Show input/output shapes
    import onnx as _onnx
    model = _onnx.load(onnx_path)
    for tensor in list(model.graph.input) + list(model.graph.output):
        dims = [d.dim_param or str(d.dim_value) for d in tensor.type.tensor_type.shape.dim]
        print(f'  {tensor.name}: [{", ".join(dims)}]')
````

### ONNX config.pbtxt

The ONNX backend configuration specifies:
- **backend:** `"onnxruntime"` for the ONNX Runtime backend
- **inputs:** `input_ids` and `attention_mask` as INT64 tensors with dynamic sequence length (`dims: [-1]`)
- **outputs:** `logits` as FP32 with 2 classes (POSITIVE/NEGATIVE)
- **dynamic_batching:** Triton accumulates individual requests into batches server-side

> The batch dimension is NOT included in `dims` — Triton adds it automatically based on `max_batch_size`.

````python
config_text = '''name: "distilbert_onnx"
backend: "onnxruntime"
max_batch_size: 8

input [
  {
    name: "input_ids"
    data_type: TYPE_INT64
    dims: [-1]
  },
  {
    name: "attention_mask"
    data_type: TYPE_INT64
    dims: [-1]
  }
]
output [
  {
    name: "logits"
    data_type: TYPE_FP32
    dims: [2]
  }
]

instance_group [
  {
    count: 1
    kind: KIND_CPU
  }
]

dynamic_batching {
  preferred_batch_size: [2, 4, 8]
  max_queue_delay_microseconds: 100000
}'''

with open(f'{MODEL_REPO}/distilbert_onnx/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/distilbert_onnx/config.pbtxt')
````

````python
config_text = '''name: "bert_onnx"
backend: "onnxruntime"
max_batch_size: 8

input [
  {
    name: "input_ids"
    data_type: TYPE_INT64
    dims: [-1]
  },
  {
    name: "attention_mask"
    data_type: TYPE_INT64
    dims: [-1]
  }
]
output [
  {
    name: "logits"
    data_type: TYPE_FP32
    dims: [2]
  }
]

instance_group [
  {
    count: 1
    kind: KIND_CPU
  }
]

dynamic_batching {
  preferred_batch_size: [2, 4, 8]
  max_queue_delay_microseconds: 100000
}'''

with open(f'{MODEL_REPO}/bert_onnx/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/bert_onnx/config.pbtxt')
````

### Model Repository

The model repository now contains three models — one Python backend and two ONNX backends:

````python
for root, dirs, files in sorted(os.walk(MODEL_REPO)):
    # Skip HF model files and __pycache__ for cleaner output
    rel = os.path.relpath(root, MODEL_REPO)
    if 'hf_model' in rel and rel.count(os.sep) > 1:
        continue
    if '__pycache__' in rel:
        continue
    level = rel.count(os.sep)
    indent = '  ' * level
    dirname = os.path.basename(root)
    if rel == '.':
        print(f'model_repository/')
    elif 'hf_model' in dirname:
        print(f'{indent}{dirname}/  (HuggingFace model files)')
    else:
        print(f'{indent}{dirname}/')
    if 'hf_model' not in rel:
        subindent = '  ' * (level + 1)
        for f in sorted(files):
            size = os.path.getsize(os.path.join(root, f))
            if size > 1024 * 1024:
                print(f'{subindent}{f}  ({size / 1024 / 1024:.1f} MB)')
            else:
                print(f'{subindent}{f}')
````

## Launch Triton

Start Triton Inference Server with the model repository mounted. Triton exposes three ports:
- **8000** — HTTP inference API
- **8001** — gRPC inference API
- **8002** — Prometheus metrics

### Inference: Python Backend

Send raw text to the Python backend model. It handles tokenization internally via `transformers.pipeline`.

````python
TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]

# Prepare input: TEXT as BYTES (strings) with shape [batch, 1]
text_array = np.array([[t] for t in TEST_INSTANCES], dtype=object)
input_tensor = httpclient.InferInput('TEXT', list(text_array.shape), 'BYTES')
input_tensor.set_data_from_numpy(text_array)

# Request both outputs
output_label = httpclient.InferRequestedOutput('LABEL')
output_score = httpclient.InferRequestedOutput('SCORE')

# Inference
result = client.infer('distilbert_python', [input_tensor], outputs=[output_label, output_score])
labels = result.as_numpy('LABEL')
scores = result.as_numpy('SCORE')

print('Python backend (distilbert_python) — raw text in, labels out:\n')
for text, label, score in zip(TEST_INSTANCES, labels.flatten(), scores.flatten()):
    print(f'  "{text[:60]}..."')
    print(f'    => {label.decode()} (score: {score:.6f})')
    print()
````

### Inference: ONNX Backend

ONNX models take pre-tokenized inputs — the client must tokenize text into `input_ids` and `attention_mask` tensors before sending. The model returns raw logits that the client converts to labels and probabilities.

This client-side preprocessing is a key trade-off: ONNX is faster but shifts tokenization work to the client. The [Pipelines and Ensembles notebook](./Triton%20Inference%20Server%20-%20Pipelines%20and%20Ensembles.ipynb) solves this with two approaches:
- **Ensemble models** — chain a Python tokenizer → ONNX model → Python postprocessor declaratively in `config.pbtxt`
- **Business Logic Scripting (BLS)** — chain the same stages in Python code, with conditional routing and dynamic model selection

````python
# Load tokenizer for client-side preprocessing
tokenizer = AutoTokenizer.from_pretrained(f'{MODELS_LOCAL}/v1')

# Tokenize
encoded = tokenizer(
    TEST_INSTANCES, padding=True, truncation=True, max_length=128, return_tensors='np',
)

# Prepare inputs
input_ids = httpclient.InferInput('input_ids', list(encoded['input_ids'].shape), 'INT64')
input_ids.set_data_from_numpy(encoded['input_ids'].astype(np.int64))

attention_mask = httpclient.InferInput('attention_mask', list(encoded['attention_mask'].shape), 'INT64')
attention_mask.set_data_from_numpy(encoded['attention_mask'].astype(np.int64))

output_logits = httpclient.InferRequestedOutput('logits')

# Inference
result = client.infer('distilbert_onnx', [input_ids, attention_mask], outputs=[output_logits])
logits = result.as_numpy('logits')

# Postprocess: softmax + label mapping
def softmax(x, axis=1):
    e_x = np.exp(x - np.max(x, axis=axis, keepdims=True))
    return e_x / e_x.sum(axis=axis, keepdims=True)

probs = softmax(logits)
id2label = {0: 'NEGATIVE', 1: 'POSITIVE'}

print('ONNX backend (distilbert_onnx) — tokenized input, raw logits out:\n')
for text, prob in zip(TEST_INSTANCES, probs):
    idx = prob.argmax()
    print(f'  "{text[:60]}..."')
    print(f'    => {id2label[idx]} (score: {prob[idx]:.6f})')
    print()
````

## Multi-Model Serving

Triton serves all models in the repository simultaneously. Route requests to different models by specifying the model name in `client.infer(model_name, ...)` — no separate containers or endpoints needed.

> **No "default model":** On Vertex AI Endpoints, you deploy models with a traffic split and `endpoint.predict()` routes to the default without specifying a model. Triton works differently — every inference request must include the model name. There is no traffic-based routing or default. The client always chooses which model to call.

Here we compare DistilBERT and BERT ONNX models on the same inputs. Notice the different label formats: DistilBERT uses `POSITIVE`/`NEGATIVE` while BERT uses `LABEL_0`/`LABEL_1` (reflecting their different `config.json` label mappings).

> **Fan-out (one input → multiple models):** This section sends the same input to multiple models via separate client calls. For server-side fan-out — where Triton automatically routes one input through multiple models and returns combined results — see the ensemble and BLS patterns in the [Pipelines and Ensembles notebook](./Triton%20Inference%20Server%20-%20Pipelines%20and%20Ensembles.ipynb).

````python
# Labels for each model
label_maps = {
    'distilbert_onnx': {0: 'NEGATIVE', 1: 'POSITIVE'},
    'bert_onnx': {0: 'LABEL_0', 1: 'LABEL_1'},
}

for model_name, id2label in label_maps.items():
    # Tokenize with the appropriate tokenizer
    version = 'v1' if 'distilbert' in model_name else 'v2'
    tok = AutoTokenizer.from_pretrained(f'{MODELS_LOCAL}/{version}')
    encoded = tok(TEST_INSTANCES, padding=True, truncation=True, max_length=128, return_tensors='np')

    input_ids = httpclient.InferInput('input_ids', list(encoded['input_ids'].shape), 'INT64')
    input_ids.set_data_from_numpy(encoded['input_ids'].astype(np.int64))
    attention_mask = httpclient.InferInput('attention_mask', list(encoded['attention_mask'].shape), 'INT64')
    attention_mask.set_data_from_numpy(encoded['attention_mask'].astype(np.int64))

    result = client.infer(model_name, [input_ids, attention_mask])
    logits = result.as_numpy('logits')
    probs = softmax(logits)

    print(f'\n=== {model_name} ===')
    for text, prob in zip(TEST_INSTANCES, probs):
        idx = prob.argmax()
        print(f'  "{text[:50]}..." => {id2label[idx]} ({prob[idx]:.4f})')
````

## Performance Comparison

Compare latency across backends and protocols. On CPU with small batches, the Python backend can be faster — PyTorch's optimized CPU kernels and dynamic sequence padding outperform ONNX Runtime's overhead. ONNX's advantage appears with GPU acceleration, TensorRT optimization, and concurrent multi-instance serving where it avoids the Python GIL.

> **Note:** These are single-request latencies on CPU. Production deployments with dynamic batching and GPU would show different characteristics — see the [Triton on GKE](./Triton%20on%20GKE.ipynb) notebook for GPU benchmarks.

````python
import concurrent.futures

def benchmark(fn, name, n=20):
    """Run fn n times, return median latency in ms."""
    times = []
    for _ in range(n):
        start = time.time()
        fn()
        times.append((time.time() - start) * 1000)
    times.sort()
    median = times[len(times) // 2]
    return median

# Prepare inputs once
text_array = np.array([[t] for t in TEST_INSTANCES], dtype=object)
tok = AutoTokenizer.from_pretrained(f'{MODELS_LOCAL}/v1')
encoded = tok(TEST_INSTANCES, padding=True, truncation=True, max_length=128, return_tensors='np')

def python_http():
    inp = httpclient.InferInput('TEXT', list(text_array.shape), 'BYTES')
    inp.set_data_from_numpy(text_array)
    client.infer('distilbert_python', [inp])

def onnx_http():
    i1 = httpclient.InferInput('input_ids', list(encoded['input_ids'].shape), 'INT64')
    i1.set_data_from_numpy(encoded['input_ids'].astype(np.int64))
    i2 = httpclient.InferInput('attention_mask', list(encoded['attention_mask'].shape), 'INT64')
    i2.set_data_from_numpy(encoded['attention_mask'].astype(np.int64))
    client.infer('distilbert_onnx', [i1, i2])

def python_grpc():
    inp = grpcclient.InferInput('TEXT', list(text_array.shape), 'BYTES')
    inp.set_data_from_numpy(text_array)
    grpc_client.infer('distilbert_python', [inp])

def onnx_grpc():
    i1 = grpcclient.InferInput('input_ids', list(encoded['input_ids'].shape), 'INT64')
    i1.set_data_from_numpy(encoded['input_ids'].astype(np.int64))
    i2 = grpcclient.InferInput('attention_mask', list(encoded['attention_mask'].shape), 'INT64')
    i2.set_data_from_numpy(encoded['attention_mask'].astype(np.int64))
    grpc_client.infer('distilbert_onnx', [i1, i2])

# Warmup
for fn in [python_http, onnx_http, python_grpc, onnx_grpc]:
    fn()

results = {
    'Python + HTTP': benchmark(python_http, 'Python+HTTP'),
    'ONNX + HTTP': benchmark(onnx_http, 'ONNX+HTTP'),
    'Python + gRPC': benchmark(python_grpc, 'Python+gRPC'),
    'ONNX + gRPC': benchmark(onnx_grpc, 'ONNX+gRPC'),
}

print(f'Latency comparison (median of 20 calls, batch of {len(TEST_INSTANCES)}):')
print(f'  {"Backend + Protocol":<20} {"Latency (ms)":>12}')
print(f'  {"-"*20} {"-"*12}')
for name, ms in results.items():
    print(f'  {name:<20} {ms:>10.1f} ms')

fastest = min(results, key=results.get)
slowest = max(results, key=results.get)
speedup = results[slowest] / results[fastest]
print(f'\n{fastest} is {speedup:.1f}x faster than {slowest}')
````

## Model Versioning

Triton supports multiple versions of the same model via numbered subdirectories. The **version policy** in `config.pbtxt` controls which versions are loaded:

```protobuf
# Load only the latest (highest-numbered) version (default)
version_policy { latest { num_versions: 1 } }

# Load all versions
version_policy { all {} }

# Load specific versions
version_policy { specific { versions: [1, 3] } }
```

When a client sends an inference request, it can optionally specify a version. Without a version, Triton routes to the latest loaded version.

**Version directories:**
```
sentiment_model/
├── config.pbtxt
├── 1/                    ← version 1 (e.g., initial training)
│   └── model.onnx
├── 2/                    ← version 2 (e.g., retrained on new data)
│   └── model.onnx
└── 3/                    ← version 3 (e.g., fine-tuned)
    └── model.onnx
```

This is different from multi-model serving (previous section), where different models have separate directories. Versioning is for tracking iterations of the **same** model.

````python
# Query model configuration to see current version policy
config = client.get_model_config('distilbert_onnx')
print(f'distilbert_onnx config:')
print(f'  max_batch_size: {config.get("max_batch_size", "N/A")}')
print(f'  version_policy: {config.get("version_policy", "latest (default)")}')
print(f'  dynamic_batching: {config.get("dynamic_batching", {})}')

# Check which versions are available
meta = client.get_model_metadata('distilbert_onnx')
print(f'\n  loaded versions: {meta.get("versions", ["unknown"])}')
````

````python
# Demonstrate version routing
tok = AutoTokenizer.from_pretrained(f'{MODELS_LOCAL}/v1')
test_text = "This movie is absolutely wonderful and I loved every minute of it!"
encoded = tok([test_text], padding='max_length', truncation=True, max_length=128, return_tensors='np')

def infer_version(model_version=None):
    """Send inference to distilbert_onnx with optional version."""
    i1 = httpclient.InferInput('input_ids', list(encoded['input_ids'].shape), 'INT64')
    i1.set_data_from_numpy(encoded['input_ids'].astype(np.int64))
    i2 = httpclient.InferInput('attention_mask', list(encoded['attention_mask'].shape), 'INT64')
    i2.set_data_from_numpy(encoded['attention_mask'].astype(np.int64))

    result = client.infer(
        'distilbert_onnx', [i1, i2],
        model_version=model_version if model_version else '',
    )
    logits = result.as_numpy('logits')
    probs = softmax(logits)
    return logits[0], probs[0]

# Request with no version (routes to latest = v2)
logits_default, probs_default = infer_version()
# Request version 1 explicitly
logits_v1, probs_v1 = infer_version('1')
# Request version 2 explicitly
logits_v2, probs_v2 = infer_version('2')

print(f'Input: "{test_text}"\n')
print(f'{"Request":<25} {"Logits":<30} {"Prediction":<15} {"Score"}')
print(f'{"-"*25} {"-"*30} {"-"*15} {"-"*8}')

for label, logits, probs in [
    ('No version (→ latest)', logits_default, probs_default),
    ('model_version="1"', logits_v1, probs_v1),
    ('model_version="2"', logits_v2, probs_v2),
]:
    idx = probs.argmax()
    pred = 'POSITIVE' if idx == 1 else 'NEGATIVE'
    print(f'{label:<25} [{logits[0]:>8.4f}, {logits[1]:>8.4f}]  {pred:<15} {probs[idx]:.6f}')

print(f'\nDefault (no version) matches v2: {np.allclose(logits_default, logits_v2)}')
print(f'v1 and v2 differ: {not np.allclose(logits_v1, logits_v2)}')
````

**Key takeaways:**
- Without `model_version`, Triton routes to the **latest** (highest-numbered) loaded version
- Clients can pin to a specific version with `model_version='1'` — useful for A/B testing or gradual rollouts
- `version_policy { all {} }` loads every version directory; `latest { num_versions: 1 }` (default) loads only the newest
- Version routing is per-model — different models in the repository have independent version policies

## Dynamic Batching

Dynamic batching is one of Triton's most powerful features. When multiple requests arrive at roughly the same time, Triton combines them into a single batch for inference — improving GPU/CPU utilization without requiring the client to batch requests.

Our ONNX configs already have dynamic batching enabled:

```protobuf
dynamic_batching {
  preferred_batch_size: [2, 4, 8]          # target batch sizes
  max_queue_delay_microseconds: 100000     # wait up to 100ms to form a batch
}
```

**How it works:**
1. Request A arrives → queued
2. Request B arrives 10ms later → queued
3. Triton forms batch [A, B] → single forward pass → individual responses

Let's demonstrate by sending concurrent requests and checking Triton's batch statistics.

````python
# Fresh client — avoids stale gevent state if a previous cell was interrupted
client = httpclient.InferenceServerClient(url='localhost:8000')

# Load tokenizer once (not per-request)
tok = AutoTokenizer.from_pretrained(f'{MODELS_LOCAL}/v1')

def send_single_request(text, http_client=None):
    """Send a single text through the ONNX model."""
    if http_client is None:
        http_client = client
    encoded = tok([text], padding='max_length', truncation=True, max_length=128, return_tensors='np')

    i1 = httpclient.InferInput('input_ids', list(encoded['input_ids'].shape), 'INT64')
    i1.set_data_from_numpy(encoded['input_ids'].astype(np.int64))
    i2 = httpclient.InferInput('attention_mask', list(encoded['attention_mask'].shape), 'INT64')
    i2.set_data_from_numpy(encoded['attention_mask'].astype(np.int64))

    result = http_client.infer('distilbert_onnx', [i1, i2])
    return result.as_numpy('logits')

# Send 8 concurrent single-text requests
texts = [
    "Great product, highly recommend!",
    "Terrible experience, would not buy again.",
    "It was okay, nothing special.",
    "Best purchase I've ever made!",
    "Complete waste of money.",
    "Surprisingly good quality for the price.",
    "Not what I expected at all.",
    "Five stars, absolutely perfect!",
]

# Sequential baseline: send 8 requests one at a time
print(f'Sequential baseline ({len(texts)} requests) ...')
start = time.time()
for t in texts:
    send_single_request(t)
sequential_ms = (time.time() - start) * 1000
print(f'  Sequential: {sequential_ms:.0f} ms ({sequential_ms / len(texts):.0f} ms/request)')

# Concurrent: send all 8 at once — Triton batches them server-side
# Each thread gets its own InferenceServerClient to avoid gevent connection pool deadlocks
# (tritonclient.http uses geventhttpclient, which has cooperative scheduling that conflicts
# with OS threads sharing a single client's connection pool)
print(f'\nConcurrent ({len(texts)} requests) ...')
start = time.time()
with concurrent.futures.ThreadPoolExecutor(max_workers=8) as executor:
    futures = [
        executor.submit(
            send_single_request, t,
            httpclient.InferenceServerClient(url='localhost:8000'),
        )
        for t in texts
    ]
    batch_results = [f.result() for f in futures]
concurrent_ms = (time.time() - start) * 1000
print(f'  Concurrent: {concurrent_ms:.0f} ms ({concurrent_ms / len(texts):.0f} ms/request)')

speedup = sequential_ms / concurrent_ms
print(f'\n  Speedup: {speedup:.1f}x (dynamic batching combines concurrent requests)')

# Check model statistics for batch size info
stats = client.get_inference_statistics('distilbert_onnx')
if 'model_stats' in stats and stats['model_stats']:
    model_stat = stats['model_stats'][0]
    if 'batch_stats' in model_stat:
        print(f'\nBatch statistics:')
        for bs in model_stat['batch_stats']:
            print(f'  batch_size={bs["batch_size"]}: count={bs["compute_infer"]["count"]}')
````

## Instance Groups

Instance groups control how many copies of a model Triton runs and on what hardware. Each instance can process requests independently, increasing throughput.

```protobuf
# Two instances on CPU
instance_group [
  { count: 2, kind: KIND_CPU }
]

# One instance on each of two GPUs
instance_group [
  { count: 1, kind: KIND_GPU, gpus: [0] },
  { count: 1, kind: KIND_GPU, gpus: [1] }
]

# Mixed: CPU for small requests, GPU for large batches
instance_group [
  { count: 1, kind: KIND_CPU },
  { count: 1, kind: KIND_GPU, gpus: [0] }
]
```

**When to increase instances:**
- CPU models: multiple instances use multiple cores (parallelism beyond Python GIL for ONNX/TF)
- GPU models: multiple instances share GPU memory and overlap compute with data transfer
- Python backend: limited by GIL — multiple instances run in separate processes

Our current configs use `count: 1` on CPU. For production, you'd tune this based on load testing — the [Triton on GKE](./Triton%20on%20GKE.ipynb) notebook covers this with HPA autoscaling.

## Triton vs FastAPI: When to Use Which

| Aspect | Custom FastAPI | Triton Inference Server |
|--------|:---:|:---:|
| **Setup complexity** | Low — write Python | Medium — model repo + config.pbtxt |
| **Serving code** | You own it | None (Triton provides it) |
| **Dynamic batching** | Must implement | Built-in, configurable |
| **Multi-model** | Multiple containers | One server |
| **Model versioning** | DIY (tags, paths) | Built-in directories |
| **gRPC support** | Must implement | Built-in |
| **Ensemble pipelines** | In-app code | Declarative config + BLS |
| **Custom pre/post processing** | Full flexibility | Python backend or BLS |
| **GPU optimization** | Manual (TensorRT, etc.) | Built-in TensorRT, CUDA graphs |
| **Framework support** | One per container | ONNX, TF, PyTorch, TensorRT, Python |
| **Metrics** | Must implement | Built-in Prometheus |
| **Container size** | Small (~1-2 GB) | Large (~15 GB) |

**Use FastAPI when:** you need full control over the serving logic, have simple single-model deployments, or want the smallest possible container.

**Use Triton when:** you need production features (batching, versioning, multi-model, ensembles), serve multiple frameworks, or want built-in GPU optimization.
