# Triton on Vertex AI Endpoints

Deploy a Triton Inference Server container to [Vertex AI Endpoints](https://cloud.google.com/vertex-ai/docs/predictions/overview) for managed, production-grade model serving. This notebook demonstrates:

- Packaging Triton with an ensemble pipeline as a custom serving container
- Building and pushing the container with Cloud Build
- Deploying to a dedicated public endpoint on CPU
- Using `rawPredict` (required — Triton's V2 protocol differs from Vertex AI's standard format)
- Multi-model routing on a single endpoint
- GPU deployment with L4 and optional TensorRT optimization

**Why `rawPredict`?** Vertex AI's standard `predict` method expects `{"instances": [...]}` format and parses the payload. Triton uses the [KServe V2 inference protocol](https://kserve.github.io/website/latest/modelserving/data_plane/v2_protocol/) with `{"inputs": [...]}` format. `rawPredict` bypasses Vertex AI's parsing and forwards the request directly to Triton.

**Story arc:**
1. Build Triton container with ensemble pipeline (tokenizer → ONNX → postprocessor)
2. Deploy to Vertex AI on CPU
3. Test with `rawPredict` — send raw text, get labels back
4. Route requests to different models (DistilBERT vs BERT)
5. Deploy with L4 GPU for faster inference

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params)
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params)

## Download Models and Build Model Repository

Download HuggingFace models, export to ONNX, and create the full ensemble pipeline (tokenizer → ONNX → postprocessor) that will be baked into the container.

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
        'triton_name': 'distilbert_onnx',
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
        'triton_name': 'bert_onnx',
    },
}
````

````python
for version, info in MODELS.items():
    local_dir = f'{MODELS_LOCAL}/{version}'
    if not os.path.exists(local_dir) or not any(os.scandir(local_dir)):
        print(f'Downloading {version}: {info["hub_name"]} ...')
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(local_dir)
        tokenizer.save_pretrained(local_dir)
    else:
        print(f'{version}: already downloaded')

    # Export to ONNX
    triton_name = info['triton_name']
    onnx_dir = f'{MODEL_REPO}/{triton_name}/1'
    onnx_path = f'{onnx_dir}/model.onnx'
    os.makedirs(onnx_dir, exist_ok=True)

    if not os.path.exists(onnx_path):
        print(f'  Exporting {triton_name} to ONNX ...')
        model = AutoModelForSequenceClassification.from_pretrained(local_dir)
        tokenizer = AutoTokenizer.from_pretrained(local_dir)
        model.eval()
        dummy = tokenizer('Test.', return_tensors='pt', padding='max_length', max_length=128, truncation=True)
        torch.onnx.export(
            model, (dummy['input_ids'], dummy['attention_mask']), onnx_path,
            input_names=['input_ids', 'attention_mask'], output_names=['logits'],
            dynamic_axes={'input_ids': {0: 'batch', 1: 'seq'}, 'attention_mask': {0: 'batch', 1: 'seq'}, 'logits': {0: 'batch'}},
            opset_version=17, do_constant_folding=True,
        )
        print(f'  Exported ({os.path.getsize(onnx_path) / 1024 / 1024:.1f} MB)')
    else:
        print(f'  {triton_name} ONNX already exists')
````

### Write Triton Configuration Files

Create the complete model repository with ONNX configs, tokenizer, postprocessor, and ensemble. Same structure as the [Ensembles notebook](./Triton%20Inference%20Server%20-%20Pipelines%20and%20Ensembles.ipynb), but baked into a container image.

````python
import shutil

# ONNX configs
for name in ['distilbert_onnx', 'bert_onnx']:
    config = f'''name: "{name}"
backend: "onnxruntime"
max_batch_size: 8
input [
  {{ name: "input_ids", data_type: TYPE_INT64, dims: [-1] }},
  {{ name: "attention_mask", data_type: TYPE_INT64, dims: [-1] }}
]
output [
  {{ name: "logits", data_type: TYPE_FP32, dims: [2] }}
]
instance_group [ {{ count: 1, kind: KIND_CPU }} ]
dynamic_batching {{ preferred_batch_size: [2, 4, 8], max_queue_delay_microseconds: 100000 }}'''
    with open(f'{MODEL_REPO}/{name}/config.pbtxt', 'w') as f:
        f.write(config)

# Tokenizer
os.makedirs(f'{MODEL_REPO}/tokenizer/1/hf_tokenizer', exist_ok=True)
with open(f'{MODEL_REPO}/tokenizer/config.pbtxt', 'w') as f:
    f.write('''name: "tokenizer"
backend: "python"
max_batch_size: 8
input [ { name: "TEXT", data_type: TYPE_STRING, dims: [1] } ]
output [
  { name: "input_ids", data_type: TYPE_INT64, dims: [-1] },
  { name: "attention_mask", data_type: TYPE_INT64, dims: [-1] }
]
instance_group [ { count: 1, kind: KIND_CPU } ]''')

with open(f'{MODEL_REPO}/tokenizer/1/model.py', 'w') as f:
    f.write('''import numpy as np
import triton_python_backend_utils as pb_utils
from transformers import AutoTokenizer

class TritonPythonModel:
    def initialize(self, args):
        model_dir = f"{args['model_repository']}/{args['model_version']}/hf_tokenizer"
        self.tokenizer = AutoTokenizer.from_pretrained(model_dir)

    def execute(self, requests):
        responses = []
        for request in requests:
            texts = [t.decode('utf-8') for t in pb_utils.get_input_tensor_by_name(request, 'TEXT').as_numpy().flatten()]
            encoded = self.tokenizer(texts, padding='max_length', truncation=True, max_length=128, return_tensors='np')
            responses.append(pb_utils.InferenceResponse([
                pb_utils.Tensor('input_ids', encoded['input_ids'].astype(np.int64)),
                pb_utils.Tensor('attention_mask', encoded['attention_mask'].astype(np.int64)),
            ]))
        return responses

    def finalize(self):
        pass
''')

# Copy tokenizer files
for f in Path(f'{MODELS_LOCAL}/v1').glob('*'):
    if f.name in ('tokenizer.json', 'tokenizer_config.json', 'vocab.txt', 'special_tokens_map.json'):
        shutil.copy2(f, f'{MODEL_REPO}/tokenizer/1/hf_tokenizer/')

# Postprocessor
os.makedirs(f'{MODEL_REPO}/postprocessor/1', exist_ok=True)
with open(f'{MODEL_REPO}/postprocessor/config.pbtxt', 'w') as f:
    f.write('''name: "postprocessor"
backend: "python"
max_batch_size: 8
input [ { name: "logits", data_type: TYPE_FP32, dims: [2] } ]
output [
  { name: "LABEL", data_type: TYPE_STRING, dims: [1] },
  { name: "SCORE", data_type: TYPE_FP32, dims: [1] }
]
instance_group [ { count: 1, kind: KIND_CPU } ]''')

with open(f'{MODEL_REPO}/postprocessor/1/model.py', 'w') as f:
    f.write('''import numpy as np
import triton_python_backend_utils as pb_utils

class TritonPythonModel:
    def initialize(self, args):
        self.id2label = {0: 'NEGATIVE', 1: 'POSITIVE'}

    def execute(self, requests):
        responses = []
        for request in requests:
            logits = pb_utils.get_input_tensor_by_name(request, 'logits').as_numpy()
            exp_logits = np.exp(logits - np.max(logits, axis=1, keepdims=True))
            probs = exp_logits / exp_logits.sum(axis=1, keepdims=True)
            pred = probs.argmax(axis=1)
            labels = np.array([[self.id2label[i]] for i in pred], dtype=object)
            scores = np.array([[probs[j, pred[j]]] for j in range(len(pred))], dtype=np.float32)
            responses.append(pb_utils.InferenceResponse([
                pb_utils.Tensor('LABEL', labels), pb_utils.Tensor('SCORE', scores),
            ]))
        return responses

    def finalize(self):
        pass
''')

# Ensemble (DistilBERT)
os.makedirs(f'{MODEL_REPO}/sentiment_ensemble/1', exist_ok=True)
with open(f'{MODEL_REPO}/sentiment_ensemble/config.pbtxt', 'w') as f:
    f.write('''name: "sentiment_ensemble"
platform: "ensemble"
max_batch_size: 8
input [ { name: "TEXT", data_type: TYPE_STRING, dims: [1] } ]
output [
  { name: "LABEL", data_type: TYPE_STRING, dims: [1] },
  { name: "SCORE", data_type: TYPE_FP32, dims: [1] }
]
ensemble_scheduling {
  step [
    { model_name: "tokenizer", model_version: -1,
      input_map { key: "TEXT", value: "TEXT" },
      output_map { key: "input_ids", value: "input_ids" },
      output_map { key: "attention_mask", value: "attention_mask" } },
    { model_name: "distilbert_onnx", model_version: -1,
      input_map { key: "input_ids", value: "input_ids" },
      input_map { key: "attention_mask", value: "attention_mask" },
      output_map { key: "logits", value: "logits" } },
    { model_name: "postprocessor", model_version: -1,
      input_map { key: "logits", value: "logits" },
      output_map { key: "LABEL", value: "LABEL" },
      output_map { key: "SCORE", value: "SCORE" } }
  ]
}''')

print('Model repository created:')
for root, dirs, files in sorted(os.walk(MODEL_REPO)):
    rel = os.path.relpath(root, MODEL_REPO)
    level = rel.count(os.sep) if rel != '.' else -1
    indent = '  ' * (level + 1)
    dirname = os.path.basename(root)
    if 'hf_tokenizer' in dirname:
        print(f'{indent}{dirname}/  (tokenizer files)')
        continue
    print(f'{indent}{dirname}/' if rel != '.' else 'model_repository/')
    if 'hf_tokenizer' not in rel:
        for f in sorted(files):
            size = os.path.getsize(os.path.join(root, f))
            s = f'  ({size/1024/1024:.1f} MB)' if size > 1024*1024 else ''
            print(f'{indent}  {f}{s}')
````

## Build Container with Cloud Build

Package the Triton model repository into a container image. The Dockerfile:
1. Starts from the Triton base image
2. Installs `transformers` and `sentencepiece` for the Python backend models
3. Copies the model repository into `/models`
4. Sets the entrypoint to `tritonserver` listening on port 8080 (Vertex AI's expected port)

> **Note:** The Triton base image is ~15 GB. Cloud Build takes 5-10 minutes.

````python
dockerfile_text = '''FROM nvcr.io/nvidia/tritonserver:24.12-py3

RUN pip install --no-cache-dir transformers sentencepiece

COPY model_repository /models

ENTRYPOINT ["tritonserver", \
    "--model-repository=/models", \
    "--http-port=8080", \
    "--allow-grpc=false", \
    "--allow-metrics=false", \
    "--log-verbose=0"]
'''

with open(f'{SOURCE_DIR}/Dockerfile', 'w') as f:
    f.write(dockerfile_text)
print(f'Wrote {SOURCE_DIR}/Dockerfile')
````

## Inference with rawPredict

Vertex AI's standard `predict()` method expects `{"instances": [...]}` and parses the request. Triton uses the [KServe V2 protocol](https://kserve.github.io/website/latest/modelserving/data_plane/v2_protocol/) with a different format: `{"inputs": [...]}`.

Use `rawPredict` to bypass Vertex AI's parsing and forward the request directly to Triton. The URL path after the endpoint base becomes the Triton route.

### Ensemble Inference

Send raw text to the ensemble — Triton handles tokenization, inference, and postprocessing server-side.

````python
TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]

# Build Triton V2 request body
triton_request = {
    "inputs": [
        {
            "name": "TEXT",
            "shape": [len(TEST_INSTANCES), 1],
            "datatype": "BYTES",
            "data": [[t] for t in TEST_INSTANCES],
        }
    ],
    "outputs": [
        {"name": "LABEL"},
        {"name": "SCORE"},
    ]
}

# Get access token for authentication
credentials, _ = google.auth.default()
credentials.refresh(google.auth.transport.requests.Request())

# rawPredict: POST to the Triton model's infer endpoint
# The endpoint URL includes the Triton model name in the path
raw_predict_url = f'https://{endpoint.dedicated_endpoint_dns}/v2/models/sentiment_ensemble/infer'

response = httpx.post(
    raw_predict_url,
    json=triton_request,
    headers={
        'Authorization': f'Bearer {credentials.token}',
        'Content-Type': 'application/json',
    },
    timeout=60,
)

result = response.json()

print(f'Ensemble rawPredict (status: {response.status_code}):\n')
# Parse V2 response
outputs = {o['name']: o['data'] for o in result['outputs']}
for text, label, score in zip(TEST_INSTANCES, outputs['LABEL'], outputs['SCORE']):
    if isinstance(label, list):
        label = label[0]
    if isinstance(score, list):
        score = score[0]
    print(f'  "{text[:60]}..."')
    print(f'    => {label} (score: {score:.6f})')
    print()
````

## Multi-Model Routing

Triton serves all models in the repository on the same endpoint. Route to different models by changing the model name in the URL path.

### Direct ONNX Inference

Bypass the ensemble and call the ONNX model directly (requires client-side tokenization).

````python
for model_name in ['distilbert_onnx', 'bert_onnx']:
    version = 'v1' if 'distilbert' in model_name else 'v2'
    tok = AutoTokenizer.from_pretrained(f'{MODELS_LOCAL}/{version}')
    enc = tok(TEST_INSTANCES, padding=True, truncation=True, max_length=128, return_tensors='np')

    triton_request = {
        "inputs": [
            {"name": "input_ids", "shape": list(enc['input_ids'].shape), "datatype": "INT64",
             "data": enc['input_ids'].astype(int).tolist()},
            {"name": "attention_mask", "shape": list(enc['attention_mask'].shape), "datatype": "INT64",
             "data": enc['attention_mask'].astype(int).tolist()},
        ],
        "outputs": [{"name": "logits"}],
    }

    credentials.refresh(google.auth.transport.requests.Request())
    url = f'https://{endpoint.dedicated_endpoint_dns}/v2/models/{model_name}/infer'
    resp = httpx.post(url, json=triton_request,
                      headers={'Authorization': f'Bearer {credentials.token}', 'Content-Type': 'application/json'},
                      timeout=60)
    logits = np.array(resp.json()['outputs'][0]['data']).reshape(resp.json()['outputs'][0]['shape'])

    # Postprocess
    exp_l = np.exp(logits - np.max(logits, axis=1, keepdims=True))
    probs = exp_l / exp_l.sum(axis=1, keepdims=True)
    id2label = {0: 'NEGATIVE', 1: 'POSITIVE'} if 'distilbert' in model_name else {0: 'LABEL_0', 1: 'LABEL_1'}

    print(f'\n=== {model_name} (direct ONNX) ===')
    for text, prob in zip(TEST_INSTANCES, probs):
        idx = prob.argmax()
        print(f'  "{text[:50]}..." => {id2label[idx]} ({prob[idx]:.4f})')
````

## GPU Deployment (L4)

Deploy the same Triton container with an L4 GPU. For GPU deployment, update the ONNX configs to use `KIND_GPU` instead of `KIND_CPU`. Since our model repository is baked into the container, we'll rebuild with GPU-enabled configs.

> **Note:** For production, you'd parameterize the config or use separate GPU/CPU model repositories. For this exercise, we rebuild the container.

### Update Config for GPU

Change `instance_group` to `KIND_GPU` and optionally enable TensorRT optimization:

```protobuf
instance_group [
  { count: 1, kind: KIND_GPU, gpus: [0] }
]

# Optional: TensorRT auto-optimization (first inference is slower, subsequent are faster)
optimization {
  execution_accelerators {
    gpu_execution_accelerator: [{ name: "tensorrt" }]
  }
}
```

````python
# Update ONNX configs for GPU
for name in ['distilbert_onnx', 'bert_onnx']:
    config = f'''name: "{name}"
backend: "onnxruntime"
max_batch_size: 8
input [
  {{ name: "input_ids", data_type: TYPE_INT64, dims: [-1] }},
  {{ name: "attention_mask", data_type: TYPE_INT64, dims: [-1] }}
]
output [
  {{ name: "logits", data_type: TYPE_FP32, dims: [2] }}
]
instance_group [ {{ count: 1, kind: KIND_GPU, gpus: [0] }} ]
dynamic_batching {{ preferred_batch_size: [2, 4, 8], max_queue_delay_microseconds: 100000 }}'''
    with open(f'{MODEL_REPO}/{name}/config.pbtxt', 'w') as f:
        f.write(config)

print('Updated ONNX configs to KIND_GPU')

# Re-upload source and rebuild
for root, dirs, files in os.walk(SOURCE_DIR):
    for fname in files:
        local_path = os.path.join(root, fname)
        blob_name = f'{SERIES}/{EXPERIMENT}/source/{os.path.relpath(local_path, SOURCE_DIR)}'
        bucket.blob(blob_name).upload_from_filename(local_path)

GPU_IMAGE = f'{REPOSITORY}/{EXPERIMENT}-gpu'

build = cloudbuild_v1.Build(steps=[])
build.steps.append({'name': 'gcr.io/cloud-builders/gsutil',
    'args': ['cp', '-r', f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/source/*', '/workspace']})
build.steps.append({'name': 'gcr.io/cloud-builders/docker',
    'args': ['build', '-t', GPU_IMAGE, '/workspace']})
build.images = [GPU_IMAGE]
build.options = cloudbuild_v1.BuildOptions(machine_type=cloudbuild_v1.BuildOptions.MachineType.E2_HIGHCPU_32)

print('Building GPU container ...')
operation = cb.create_build(project_id=PROJECT_ID, build=build)
build_response = operation.result()
print(f'Build status: {build_response.status.name}')
````

````python
# Upload GPU model
gpu_model = aiplatform.Model.upload(
    display_name=f'{EXPERIMENT}-triton-gpu',
    serving_container_image_uri=GPU_IMAGE,
    serving_container_ports=[8080],
    labels={'series': SERIES, 'experiment': EXPERIMENT, 'accelerator': 'l4'},
)

# Create GPU endpoint
gpu_endpoint = aiplatform.Endpoint.create(
    display_name=f'{EXPERIMENT}-gpu',
    dedicated_endpoint_enabled=True,
    labels={'series': SERIES, 'experiment': EXPERIMENT},
)

# Deploy with L4
gpu_endpoint.deploy(
    model=gpu_model,
    deployed_model_display_name='triton-sentiment-gpu',
    machine_type='g2-standard-4',
    accelerator_type='NVIDIA_L4',
    accelerator_count=1,
    min_replica_count=1,
    max_replica_count=1,
    traffic_percentage=100,
)
print(f'GPU endpoint deployed: {gpu_endpoint.resource_name}')
````

## Triton on Vertex AI vs FastAPI on Vertex AI

| Aspect | FastAPI on Vertex AI | Triton on Vertex AI |
|--------|:---:|:---:|
| **Prediction method** | `predict()` / `rawPredict` | `rawPredict` only |
| **Request format** | `{"instances": [...]}` | Triton V2: `{"inputs": [...]}` |
| **Multi-model** | One model per container | Multiple models per container |
| **Batching** | Must implement | Built-in dynamic batching |
| **Container size** | ~2 GB | ~15 GB |
| **Build time** | ~2 min | ~5-10 min |
| **GPU optimization** | Manual | Built-in TensorRT support |
| **Model versioning** | Via Model Registry versions | Built-in + Model Registry |
| **Pipeline/ensemble** | In-app code | Declarative config |

**Use FastAPI** when: you need `predict()` compatibility, simple single-model deployment, or smallest container.

**Use Triton** when: you need multi-model, ensembles, dynamic batching, or GPU optimization out of the box.
