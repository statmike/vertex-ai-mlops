# Triton on GKE

Deploy Triton Inference Server to [GKE Autopilot](https://cloud.google.com/kubernetes-engine/docs/concepts/autopilot-overview) with full Kubernetes control over the serving infrastructure. This notebook demonstrates:

- GKE Autopilot cluster with CPU deployment (GPU shown as exercise)
- Multi-model ONNX serving with dynamic batching
- Kubernetes Deployment, Service, and Horizontal Pod Autoscaler (HPA)
- Health probes using Triton's built-in `/v2/health/ready` and `/v2/health/live` endpoints
- HPA with Triton's Prometheus metrics for inference-aware autoscaling
- Production patterns: graceful shutdown, rolling updates, startup probes

**GKE vs Cloud Run vs Vertex AI for Triton:**
- GKE: Full Kubernetes control, custom metrics HPA, GCS FUSE model loading, any GPU type
- Cloud Run: Simplest deployment, scale-to-zero, limited GPU options
- Vertex AI: Managed ML platform, Model Registry, rawPredict

**Story arc:**
1. Create GKE Autopilot cluster
2. Build and deploy Triton with baked-in ONNX model repository
3. Test inference via LoadBalancer Service
4. Configure HPA with Triton metrics
5. Production patterns (probes, rolling updates)

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params)
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params)

## Build Triton Container

Download models, export to ONNX, and build via Cloud Build. The model repository contains two ONNX models. Tokenization and postprocessing happen client-side.

> **Ensemble pipelines:** For server-side tokenization via Triton's ensemble scheduler and Python backend, see [Triton Inference Server - Pipelines and Ensembles](./Triton%20Inference%20Server%20-%20Pipelines%20and%20Ensembles.ipynb) and the [Vertex AI notebook](./Triton%20on%20Vertex%20AI%20Endpoints.ipynb) (GPU deployment with full ensemble). GKE supports the Python backend — mount `/dev/shm` as an emptyDir (already configured in the deployment below) — but it is omitted here for consistency across deployment targets.

````python
from torch.export import Dim

batch = Dim('batch', min=1, max=8)
seq = Dim('seq', min=1, max=512)
dummy_texts = ['Sentence one.', 'Sentence two.', 'Sentence three.', 'Sentence four.']

MODELS = {
    'v1': {'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english', 'triton_name': 'distilbert_onnx'},
    'v2': {'hub_name': 'textattack/bert-base-uncased-SST-2', 'triton_name': 'bert_onnx'},
}

for version, info in MODELS.items():
    local_dir = f'{MODELS_LOCAL}/{version}'
    if not os.path.exists(local_dir) or not any(os.scandir(local_dir)):
        print(f'Downloading {version}: {info["hub_name"]} ...')
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(local_dir)
        tokenizer.save_pretrained(local_dir)
    else:
        print(f'{version}: already downloaded')

    onnx_dir = f'{MODEL_REPO}/{info["triton_name"]}/1'
    onnx_path = f'{onnx_dir}/model.onnx'
    os.makedirs(onnx_dir, exist_ok=True)
    if not os.path.exists(onnx_path):
        print(f'  Exporting to ONNX ...')
        m = AutoModelForSequenceClassification.from_pretrained(local_dir)
        t = AutoTokenizer.from_pretrained(local_dir)
        m.eval()
        dummy = t(
            dummy_texts, return_tensors='pt',
            padding='max_length', max_length=128, truncation=True,
        )
        torch.onnx.export(
            m, (dummy['input_ids'], dummy['attention_mask']), onnx_path,
            input_names=['input_ids', 'attention_mask'], output_names=['logits'],
            dynamic_shapes=({0: batch, 1: seq}, {0: batch, 1: seq}),
        )
        total_bytes = sum(os.path.getsize(os.path.join(onnx_dir, f)) for f in os.listdir(onnx_dir))
        print(f'  Exported ({total_bytes / 1024 / 1024:.1f} MB)')
````

````python
import shutil

# Remove Python backend models from any previous runs
for name in ['tokenizer', 'postprocessor', 'sentiment_ensemble']:
    path = f'{MODEL_REPO}/{name}'
    if os.path.exists(path):
        shutil.rmtree(path)
        print(f'Removed leftover: {name}/')

# ONNX configs (CPU)
for name in ['distilbert_onnx', 'bert_onnx']:
    config = f'''name: "{name}"
backend: "onnxruntime"
max_batch_size: 8
input [
  {{ name: "input_ids", data_type: TYPE_INT64, dims: [-1] }},
  {{ name: "attention_mask", data_type: TYPE_INT64, dims: [-1] }}
]
output [ {{ name: "logits", data_type: TYPE_FP32, dims: [2] }} ]
instance_group [ {{ count: 1, kind: KIND_CPU }} ]
dynamic_batching {{ preferred_batch_size: [2, 4, 8], max_queue_delay_microseconds: 100000 }}'''
    with open(f'{MODEL_REPO}/{name}/config.pbtxt', 'w') as f:
        f.write(config)

# Dockerfile
with open(f'{SOURCE_DIR}/Dockerfile', 'w') as f:
    f.write('''FROM nvcr.io/nvidia/tritonserver:24.12-py3

COPY model_repository /models

ENV CUDA_VISIBLE_DEVICES=""
ENTRYPOINT ["tritonserver", \
    "--model-repository=/models", \
    "--log-verbose=0"]
''')

print('Model repository and Dockerfile created.')
print(f'Models: {[d for d in os.listdir(MODEL_REPO) if os.path.isdir(f"{MODEL_REPO}/{d}")]}')
````

````python
# Artifact Registry
docker_repo = None
for repo in ar.list_repositories(parent=f'projects/{PROJECT_ID}/locations/{REGION}'):
    if AR_REPO == repo.name.split('/')[-1]:
        docker_repo = repo
if not docker_repo:
    op = ar.create_repository(request=artifactregistry_v1.CreateRepositoryRequest(
        parent=f'projects/{PROJECT_ID}/locations/{REGION}', repository_id=AR_REPO,
        repository=artifactregistry_v1.Repository(name=AR_REPO, format_=artifactregistry_v1.Repository.Format.DOCKER)))
    docker_repo = op.result()
REPOSITORY = f"{REGION}-docker.pkg.dev/{PROJECT_ID}/{docker_repo.name.split('/')[-1]}"
IMAGE_URI = f'{REPOSITORY}/{EXPERIMENT}'

# Clean up old source files in GCS (removes leftovers from previous runs)
for blob in bucket.list_blobs(prefix=f'{SERIES}/{EXPERIMENT}/source/'):
    blob.delete()

# Upload source
for root, dirs, files in os.walk(SOURCE_DIR):
    for fname in files:
        local_path = os.path.join(root, fname)
        blob_name = f'{SERIES}/{EXPERIMENT}/source/{os.path.relpath(local_path, SOURCE_DIR)}'
        bucket.blob(blob_name).upload_from_filename(local_path)

# Build
build = cloudbuild_v1.Build(steps=[
    {'name': 'gcr.io/cloud-builders/gsutil', 'args': ['cp', '-r', f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/source/*', '/workspace']},
    {'name': 'gcr.io/cloud-builders/docker', 'args': ['build', '-t', IMAGE_URI, '/workspace']},
])
build.images = [IMAGE_URI]
build.options = cloudbuild_v1.BuildOptions(machine_type=cloudbuild_v1.BuildOptions.MachineType.E2_HIGHCPU_32)

print('Starting Cloud Build ...')
op = cb.create_build(project_id=PROJECT_ID, build=build)
result = op.result()
print(f'Build status: {result.status.name}')
print(f'Image: {IMAGE_URI}')
````

## Create GKE Autopilot Cluster

Create a GKE Autopilot cluster. Autopilot manages nodes automatically — you just declare the resources your pods need and GKE provisions the right nodes.

> **Note:** Cluster creation takes ~10 minutes.

````python
cluster_path = f'projects/{PROJECT_ID}/locations/{REGION}'

# Check if cluster already exists
existing = None
try:
    existing = gke.get_cluster(name=f'{cluster_path}/clusters/{GKE_CLUSTER}')
    print(f'Cluster already exists: {existing.name}')
except Exception:
    pass

if not existing:
    print(f'Creating Autopilot cluster: {GKE_CLUSTER} ...')
    cluster = container_v1.Cluster(
        name=GKE_CLUSTER,
        autopilot=container_v1.Autopilot(enabled=True),
    )
    operation = gke.create_cluster(
        request=container_v1.CreateClusterRequest(parent=cluster_path, cluster=cluster)
    )

    # Wait for cluster creation
    while True:
        op_status = gke.get_operation(name=f'{cluster_path}/operations/{operation.name}')
        if op_status.status == container_v1.Operation.Status.DONE:
            if op_status.error and op_status.error.code:
                raise RuntimeError(f'Cluster creation failed: {op_status.error.message}')
            print('Cluster created!')
            break
        if op_status.status == container_v1.Operation.Status.ABORTING:
            raise RuntimeError(f'Cluster creation aborted: {op_status.status_message}')
        print(f'  Status: {op_status.status.name} ...')
        time.sleep(30)
````

````python
# Get cluster credentials for kubectl
cluster_info = gke.get_cluster(name=f'{cluster_path}/clusters/{GKE_CLUSTER}')
print(f'Cluster endpoint: {cluster_info.endpoint}')

# Configure kubernetes client
import google.auth
import google.auth.transport.requests
import base64, tempfile

credentials, _ = google.auth.default()
credentials.refresh(google.auth.transport.requests.Request())

# Write cluster CA cert to temp file
ca_cert = base64.b64decode(cluster_info.master_auth.cluster_ca_certificate)
ca_file = tempfile.NamedTemporaryFile(delete=False, suffix='.crt')
ca_file.write(ca_cert)
ca_file.close()

configuration = k8s_client.Configuration()
configuration.host = f'https://{cluster_info.endpoint}'
configuration.ssl_ca_cert = ca_file.name
configuration.api_key = {'authorization': f'Bearer {credentials.token}'}

api_client = k8s_client.ApiClient(configuration)
apps_v1 = k8s_client.AppsV1Api(api_client)
core_v1 = k8s_client.CoreV1Api(api_client)
autoscaling_v1 = k8s_client.AutoscalingV1Api(api_client)
print('Kubernetes client configured.')

def refresh_k8s_auth():
    """Refresh the OAuth2 token if it has expired (tokens last ~60 min)."""
    credentials.refresh(google.auth.transport.requests.Request())
    configuration.api_key = {'authorization': f'Bearer {credentials.token}'}
    print('K8s auth token refreshed.')
````

## Deploy: Baked-In Model Repository

Deploy Triton with the model repository baked into the container image. This is the simplest approach — no external storage needed. The trade-off is that updating models requires rebuilding the container.

### Deployment on CPU

````python
DEPLOYMENT_NAME = 'triton-sentiment'

deployment = k8s_client.V1Deployment(
    metadata=k8s_client.V1ObjectMeta(name=DEPLOYMENT_NAME, labels={'app': DEPLOYMENT_NAME}),
    spec=k8s_client.V1DeploymentSpec(
        replicas=1,
        selector=k8s_client.V1LabelSelector(match_labels={'app': DEPLOYMENT_NAME}),
        template=k8s_client.V1PodTemplateSpec(
            metadata=k8s_client.V1ObjectMeta(labels={'app': DEPLOYMENT_NAME}),
            spec=k8s_client.V1PodSpec(
                containers=[
                    k8s_client.V1Container(
                        name='triton',
                        image=IMAGE_URI,
                        ports=[
                            k8s_client.V1ContainerPort(container_port=8000, name='http'),
                            k8s_client.V1ContainerPort(container_port=8001, name='grpc'),
                            k8s_client.V1ContainerPort(container_port=8002, name='metrics'),
                        ],
                        resources=k8s_client.V1ResourceRequirements(
                            limits={'cpu': '4', 'memory': '16Gi'},
                            requests={'cpu': '4', 'memory': '16Gi'},
                        ),
                        volume_mounts=[
                            k8s_client.V1VolumeMount(name='dshm', mount_path='/dev/shm'),
                        ],
                        startup_probe=k8s_client.V1Probe(
                            http_get=k8s_client.V1HTTPGetAction(path='/v2/health/ready', port=8000),
                            initial_delay_seconds=30,
                            period_seconds=10,
                            failure_threshold=30,
                        ),
                        readiness_probe=k8s_client.V1Probe(
                            http_get=k8s_client.V1HTTPGetAction(path='/v2/health/ready', port=8000),
                            period_seconds=10,
                        ),
                        liveness_probe=k8s_client.V1Probe(
                            http_get=k8s_client.V1HTTPGetAction(path='/v2/health/live', port=8000),
                            period_seconds=30,
                        ),
                    ),
                ],
                volumes=[
                    k8s_client.V1Volume(
                        name='dshm',
                        empty_dir=k8s_client.V1EmptyDirVolumeSource(medium='Memory'),
                    ),
                ],
            ),
        ),
    ),
)

try:
    apps_v1.delete_namespaced_deployment(DEPLOYMENT_NAME, NAMESPACE)
    time.sleep(5)
except Exception:
    pass

apps_v1.create_namespaced_deployment(namespace=NAMESPACE, body=deployment)
print(f'Deployment {DEPLOYMENT_NAME} created.')
````

### Service (LoadBalancer)

````python
SERVICE_NAME = 'triton-service'

service = k8s_client.V1Service(
    metadata=k8s_client.V1ObjectMeta(name=SERVICE_NAME),
    spec=k8s_client.V1ServiceSpec(
        type='LoadBalancer',
        selector={'app': DEPLOYMENT_NAME},
        ports=[
            k8s_client.V1ServicePort(name='http', port=8000, target_port=8000),
            k8s_client.V1ServicePort(name='grpc', port=8001, target_port=8001),
            k8s_client.V1ServicePort(name='metrics', port=8002, target_port=8002),
        ],
    ),
)

try:
    core_v1.delete_namespaced_service(SERVICE_NAME, NAMESPACE)
    # Wait for deletion to propagate
    for _ in range(12):
        try:
            core_v1.read_namespaced_service(SERVICE_NAME, NAMESPACE)
            time.sleep(5)
        except Exception:
            break
except Exception:
    pass

try:
    core_v1.create_namespaced_service(namespace=NAMESPACE, body=service)
    print(f'Service {SERVICE_NAME} created.')
except Exception as e:
    if '409' in str(e) or 'AlreadyExists' in str(e):
        print(f'Service {SERVICE_NAME} already exists — reusing.')
    else:
        raise
````

````python
# Wait for deployment to be ready
# Image pull + model loading can take 5-15 minutes on first deploy (15 GB base image)
print('Waiting for Triton pods to be ready ...')
print('(15 GB image pull + model loading — this can take 5-15 min)')
for i in range(180):
    try:
        dep = apps_v1.read_namespaced_deployment(DEPLOYMENT_NAME, NAMESPACE)
        ready = dep.status.ready_replicas or 0
        desired = dep.status.replicas or 1
        if ready >= desired:
            print(f'Deployment ready: {ready}/{desired} replicas')
            break
        print(f'  {ready}/{desired} ready ... ({i*10}s)')
    except Exception as e:
        if '401' in str(e):
            refresh_k8s_auth()
        else:
            print(f'  Waiting ... ({i*10}s)')
    time.sleep(10)
else:
    print('Timeout waiting for deployment. Check pod status with:')
    print(f'  kubectl get pods -l app={DEPLOYMENT_NAME}')

# Wait for external IP
print('\nWaiting for external IP ...')
EXTERNAL_IP = None
for i in range(30):
    try:
        svc = core_v1.read_namespaced_service(SERVICE_NAME, NAMESPACE)
        ingress = svc.status.load_balancer.ingress
        if ingress and ingress[0].ip:
            EXTERNAL_IP = ingress[0].ip
            print(f'External IP: {EXTERNAL_IP}')
            break
    except Exception as e:
        if '401' in str(e):
            refresh_k8s_auth()
    time.sleep(10)
else:
    print('Timeout waiting for external IP.')

# Wait for LoadBalancer to be routable (IP assigned != traffic flowing)
if EXTERNAL_IP:
    print('\nWaiting for LoadBalancer to accept traffic ...')
    for i in range(30):
        try:
            resp = httpx.get(f'http://{EXTERNAL_IP}:8000/v2/health/ready', timeout=5)
            if resp.status_code == 200:
                print('Triton is reachable and ready!')
                break
        except Exception:
            pass
        if i % 3 == 0:
            print(f'  Not yet routable ... ({i*10}s)')
        time.sleep(10)
    else:
        print('Timeout waiting for LoadBalancer routing.')
````

## Test Inference

Send requests directly to Triton's HTTP API via the LoadBalancer IP — no authentication wrapper needed (unlike Vertex AI or Cloud Run). Tokenization happens client-side; the server handles ONNX inference with dynamic batching.

````python
TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]

# Client-side tokenization
tok = AutoTokenizer.from_pretrained(f'{MODELS_LOCAL}/v1')
enc = tok(TEST_INSTANCES, padding=True, truncation=True, max_length=128, return_tensors='np')

triton_request = {
    "inputs": [
        {"name": "input_ids", "shape": list(enc['input_ids'].shape), "datatype": "INT64",
         "data": enc['input_ids'].astype(int).tolist()},
        {"name": "attention_mask", "shape": list(enc['attention_mask'].shape), "datatype": "INT64",
         "data": enc['attention_mask'].astype(int).tolist()},
    ],
    "outputs": [{"name": "logits"}],
}

response = httpx.post(f'http://{EXTERNAL_IP}:8000/v2/models/distilbert_onnx/infer',
                      json=triton_request, timeout=60)
result = response.json()
logits = np.array(result['outputs'][0]['data']).reshape(result['outputs'][0]['shape'])

# Client-side postprocessing
exp_l = np.exp(logits - np.max(logits, axis=1, keepdims=True))
probs = exp_l / exp_l.sum(axis=1, keepdims=True)
id2label = {0: 'NEGATIVE', 1: 'POSITIVE'}

print(f'GKE CPU — distilbert_onnx (status: {response.status_code}):\n')
for text, prob in zip(TEST_INSTANCES, probs):
    idx = prob.argmax()
    print(f'  "{text[:60]}..."')
    print(f'    => {id2label[idx]} (score: {prob[idx]:.6f})')
    print()
````

## Model Repository Strategies

| Strategy | Pros | Cons | Best for |
|----------|------|------|----------|
| **Baked-in** (above) | Simplest, no runtime deps | Rebuild container to update models | Stable models, CI/CD pipelines |
| **GCS FUSE** | No rebuild, hot-reload capable | Startup latency (download), GCS costs | Frequent model updates |
| **PVC** | Fastest startup (pre-loaded) | Provisioning complexity | Large models, lowest latency |

### GCS FUSE Volume Mount

Mount a GCS bucket as a local filesystem using [GCS FUSE](https://cloud.google.com/kubernetes-engine/docs/how-to/persistent-volumes/cloud-storage-fuse-csi-driver). Triton reads the model repository from the mounted path at startup. Model updates in GCS are visible without rebuilding the container.

> **Note:** GCS FUSE deployment is shown as a reference pattern. To use it, upload your model repository to GCS and modify the Deployment to use the FUSE sidecar annotation.

```yaml
# Key annotations and volume config for GCS FUSE:
metadata:
  annotations:
    gke-gcsfuse/volumes: "true"        # Enable FUSE sidecar
spec:
  volumes:
    - name: model-repo
      csi:
        driver: gcsfuse.csi.storage.gke.io
        volumeAttributes:
          bucketName: "your-bucket"
          mountOptions: "implicit-dirs"
  containers:
    - name: triton
      command: ["tritonserver", "--model-repository=/models"]
      volumeMounts:
        - name: model-repo
          mountPath: /models
          subPath: "path/to/model_repository"
```

## HPA with Triton Metrics

Triton exposes [Prometheus metrics](https://docs.nvidia.com/deeplearning/triton-inference-server/user-guide/docs/user_guide/metrics.html) on port 8002. Key metrics for autoscaling:

- `nv_inference_queue_duration_us` — Time requests wait in the queue (indicates load)
- `nv_inference_request_success` — Successful inference count (throughput)
- `nv_inference_compute_infer_duration_us` — Inference compute time

For this notebook, we use a simple CPU-based HPA. Production deployments would use [custom metrics](https://cloud.google.com/kubernetes-engine/docs/how-to/custom-metrics) from Triton's Prometheus endpoint.

### Basic HPA

````python
hpa = k8s_client.V1HorizontalPodAutoscaler(
    metadata=k8s_client.V1ObjectMeta(name='triton-hpa'),
    spec=k8s_client.V1HorizontalPodAutoscalerSpec(
        scale_target_ref=k8s_client.V1CrossVersionObjectReference(
            api_version='apps/v1', kind='Deployment', name=DEPLOYMENT_NAME,
        ),
        min_replicas=1,
        max_replicas=3,
        target_cpu_utilization_percentage=70,
    ),
)

try:
    autoscaling_v1.delete_namespaced_horizontal_pod_autoscaler('triton-hpa', NAMESPACE)
except Exception:
    pass

autoscaling_v1.create_namespaced_horizontal_pod_autoscaler(namespace=NAMESPACE, body=hpa)
print('HPA created: min=1, max=3, target CPU=70%')

# Brief wait for the controller to initialize
time.sleep(5)
hpa_status = autoscaling_v1.read_namespaced_horizontal_pod_autoscaler('triton-hpa', NAMESPACE)
print(f'Current replicas: {hpa_status.status.current_replicas}')
print(f'Desired replicas: {hpa_status.status.desired_replicas}')
````

## Production Patterns

### Health Probes

Triton provides built-in health endpoints:
- **`/v2/health/live`** — Is the server process running? (liveness probe)
- **`/v2/health/ready`** — Are all models loaded and ready? (readiness probe)

Our deployment already uses these probes. The key settings:
- **startupProbe:** Higher `failureThreshold` (30) to allow time for model loading
- **readinessProbe:** Moderate frequency (10s) — removes pod from service if models fail
- **livenessProbe:** Lower frequency (30s) — restarts pod only if server process dies

### Graceful Shutdown

Triton handles SIGTERM gracefully — it stops accepting new requests, finishes in-flight requests, and exits. Kubernetes sends SIGTERM during pod termination. Set `terminationGracePeriodSeconds` to allow time for in-flight requests to complete.

### Rolling Updates

The default Deployment strategy (`RollingUpdate`) works well with Triton:
- New pods start and pass startup probe before receiving traffic
- Old pods are drained and terminated
- Zero-downtime model updates by changing the container image

## GPU Deployment (Exercise)

To deploy Triton on GKE with an L4 GPU for faster ONNX inference, make these changes:

**1. ONNX configs** — switch from `KIND_CPU` to `KIND_GPU`:

```protobuf
instance_group [ { count: 1, kind: KIND_GPU, gpus: [0] } ]
```

**2. Rebuild the container** — since the model repository is baked in, rebuild with the GPU configs:

```python
# Update ONNX configs for GPU
for name in ['distilbert_onnx', 'bert_onnx']:
    config = f'''name: "{name}"
backend: "onnxruntime"
max_batch_size: 8
input [
  {{ name: "input_ids", data_type: TYPE_INT64, dims: [-1] }},
  {{ name: "attention_mask", data_type: TYPE_INT64, dims: [-1] }}
]
output [ {{ name: "logits", data_type: TYPE_FP32, dims: [2] }} ]
instance_group [ {{ count: 1, kind: KIND_GPU, gpus: [0] }} ]
dynamic_batching {{ preferred_batch_size: [2, 4, 8], max_queue_delay_microseconds: 100000 }}'''
    with open(f'{MODEL_REPO}/{name}/config.pbtxt', 'w') as f:
        f.write(config)

# Re-upload source and rebuild
# ... (same Cloud Build pattern as above)
```

**3. Deployment resources** — add GPU to container limits and set the node selector:

```python
resources=k8s_client.V1ResourceRequirements(
    limits={'nvidia.com/gpu': '1', 'cpu': '4', 'memory': '16Gi'},
    requests={'nvidia.com/gpu': '1', 'cpu': '4', 'memory': '16Gi'},
),
# ...
node_selector={'cloud.google.com/gke-accelerator': 'nvidia-l4'},
```

**4. GPU quota** — GPU deployment requires GKE GPU quota for `nvidia-l4` in your region. GPU node provisioning can take an additional 5-15 minutes after the deployment is created.

> **Optional: TensorRT optimization** — Add to the ONNX config for automatic TensorRT conversion (first inference is slower, subsequent are faster):
> ```protobuf

>     gpu_execution_accelerator: [{ name: "tensorrt" }]
>   }

> ```
