# Triton Inference Server — Pipelines and Ensembles

Chain tokenization, inference, and postprocessing into server-side pipelines — so clients send raw text and receive human-readable results without any preprocessing code.

The [Fundamentals notebook](./Triton%20Inference%20Server%20-%20Fundamentals.ipynb) showed two trade-offs:
- **Python backend:** Accepts raw text (easy for clients) but runs in Python (slower)
- **ONNX backend:** Fast C++ inference but requires pre-tokenized input (harder for clients)

This notebook combines the best of both using Triton's two pipeline mechanisms — **ensembles** (declarative DAGs) and **BLS** (imperative Python) — across four increasingly sophisticated patterns:

### Four Pipeline Patterns

| # | Pattern | Mechanism | What It Does |
|---|---------|-----------|--------------|
| 1 | **Linear Ensemble** | Ensemble | `text → tokenizer → model → postprocessor → labels` — the basic chain |
| 2 | **Fan-Out Ensemble** | Ensemble | Same input through **two models in parallel**, results combined |
| 3 | **Routing BLS** | BLS | Routes to different models based on **input properties** (text length) |
| 4 | **Cascade BLS** | BLS | Runs fast model first, **escalates to large model** only when confidence is low |

```
Pattern 1: Linear               Pattern 2: Fan-Out
─────────────────────            ─────────────────────────────
text → tok → model → post       text → tok ─┬→ model_A → post_A ─┐
                                             └→ model_B → post_B ─┴→ combiner

Pattern 3: Routing               Pattern 4: Cascade
─────────────────────             ─────────────────────────────────
              ┌→ model_A ─┐                    ┌─ confidence ≥ 0.95? → done
text → tok ──┤            ├→ post      text → tok → fast_model → post ─┤
              └→ model_B ─┘                    └─ confidence < 0.95? → large_model → post
       (pick one by input)                           (escalate if uncertain)
```

## Download Models and Export to ONNX

Download both HuggingFace sentiment models and export them to ONNX format. The ONNX models will serve as the inference backend in our ensemble pipeline.

````python
MODELS = {
    'v1': {
        'hub_name': 'distilbert-base-uncased-finetuned-sst-2-english',
        'display_name': 'distilbert-sst2',
        'triton_name': 'distilbert_onnx',
        'id2label': {0: 'NEGATIVE', 1: 'POSITIVE'},
    },
    'v2': {
        'hub_name': 'textattack/bert-base-uncased-SST-2',
        'display_name': 'bert-sst2',
        'triton_name': 'bert_onnx',
        'id2label': {0: 'LABEL_0', 1: 'LABEL_1'},
    },
}
````

````python
for version, info in MODELS.items():
    local_dir = f'{MODELS_LOCAL}/{version}'
    if os.path.exists(local_dir) and any(os.scandir(local_dir)):
        print(f'{version} ({info["hub_name"]}) already downloaded')
        continue

    print(f'Downloading {version}: {info["hub_name"]} ...')
    model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
    tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
    model.save_pretrained(local_dir)
    tokenizer.save_pretrained(local_dir)
    print(f'  Saved to {local_dir}')
````

````python
from torch.export import Dim

batch = Dim('batch', min=1, max=8)
seq = Dim('seq', min=1, max=512)

# Multi-sample dummy input — prevents the dynamo exporter from specializing batch to 1
dummy_texts = ['Sentence one.', 'Sentence two.', 'Sentence three.', 'Sentence four.']

for version, info in MODELS.items():
    triton_name = info['triton_name']
    onnx_dir = f'{MODEL_REPO}/{triton_name}/1'
    onnx_path = f'{onnx_dir}/model.onnx'
    os.makedirs(onnx_dir, exist_ok=True)

    if os.path.exists(onnx_path):
        total_bytes = sum(os.path.getsize(os.path.join(onnx_dir, f)) for f in os.listdir(onnx_dir))
        print(f'{triton_name}: ONNX already exists ({total_bytes / 1024 / 1024:.1f} MB)')
        continue

    local_dir = f'{MODELS_LOCAL}/{version}'
    print(f'Exporting {triton_name} to ONNX ...')
    model = AutoModelForSequenceClassification.from_pretrained(local_dir)
    tokenizer = AutoTokenizer.from_pretrained(local_dir)
    model.eval()

    dummy = tokenizer(
        dummy_texts, return_tensors='pt',
        padding='max_length', max_length=128, truncation=True,
    )

    torch.onnx.export(
        model, (dummy['input_ids'], dummy['attention_mask']), onnx_path,
        input_names=['input_ids', 'attention_mask'], output_names=['logits'],
        dynamic_shapes=({0: batch, 1: seq}, {0: batch, 1: seq}),
    )

    total_bytes = sum(os.path.getsize(os.path.join(onnx_dir, f)) for f in os.listdir(onnx_dir))
    print(f'  Exported ({total_bytes / 1024 / 1024:.1f} MB)')
````

````python
for version, info in MODELS.items():
    triton_name = info['triton_name']
    config_text = f'''name: "{triton_name}"
backend: "onnxruntime"
max_batch_size: 8

input [
  {{ name: "input_ids", data_type: TYPE_INT64, dims: [-1] }},
  {{ name: "attention_mask", data_type: TYPE_INT64, dims: [-1] }}
]
output [
  {{ name: "logits", data_type: TYPE_FP32, dims: [2] }}
]

instance_group [
  {{ count: 1, kind: KIND_CPU }}
]'''

    with open(f'{MODEL_REPO}/{triton_name}/config.pbtxt', 'w') as f:
        f.write(config_text)
    print(f'Wrote {MODEL_REPO}/{triton_name}/config.pbtxt')
````

## The Problem: Client-Side Preprocessing

In the [Fundamentals notebook](./Triton%20Inference%20Server%20-%20Fundamentals.ipynb), ONNX inference required the client to:

1. **Tokenize** text into `input_ids` and `attention_mask` tensors
2. **Send** tokenized tensors to Triton
3. **Postprocess** raw logits into labels and scores

This pushes complexity to every client. Different languages, platforms, and teams must all implement the same tokenization — a source of bugs and inconsistency.

**Solution:** Move tokenization and postprocessing inside Triton as separate models, then chain them with the ONNX model in a pipeline:

```
Client sends:     "This movie is great!"
                         │
               ┌─────────▼─────────┐
               │    tokenizer       │  Python backend
               │  (AutoTokenizer)   │
               └─────────┬─────────┘
                         │ input_ids, attention_mask
               ┌─────────▼─────────┐
               │   distilbert_onnx  │  ONNX backend
               │   (inference)      │
               └─────────┬─────────┘
                         │ logits
               ┌─────────▼─────────┐
               │   postprocessor    │  Python backend
               │  (softmax + label) │
               └─────────┬─────────┘
                         │
Client receives:  POSITIVE (0.9999)
```

## Tokenizer Model (Python Backend)

The tokenizer model wraps `AutoTokenizer` from HuggingFace. It accepts raw text and outputs `input_ids` and `attention_mask` tensors ready for the ONNX model.

### config.pbtxt

````python
os.makedirs(f'{MODEL_REPO}/tokenizer/1', exist_ok=True)

config_text = '''name: "tokenizer"
backend: "python"
max_batch_size: 8

input [
  {
    name: "TEXT"
    data_type: TYPE_STRING
    dims: [1]
  }
]
output [
  {
    name: "input_ids"
    data_type: TYPE_INT64
    dims: [-1]
  },
  {
    name: "attention_mask"
    data_type: TYPE_INT64
    dims: [-1]
  }
]

instance_group [
  { count: 1, kind: KIND_CPU }
]'''

with open(f'{MODEL_REPO}/tokenizer/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/tokenizer/config.pbtxt')
````

### model.py

The tokenizer loads the DistilBERT tokenizer and pads all sequences to 128 tokens for consistent tensor shapes.

````python
model_py = '''import numpy as np
import triton_python_backend_utils as pb_utils
from transformers import AutoTokenizer


class TritonPythonModel:
    def initialize(self, args):
        model_dir = f"{args['model_repository']}/{args['model_version']}/hf_tokenizer"
        self.tokenizer = AutoTokenizer.from_pretrained(model_dir)

    def execute(self, requests):
        responses = []
        for request in requests:
            input_tensor = pb_utils.get_input_tensor_by_name(request, 'TEXT')
            texts = [t.decode('utf-8') for t in input_tensor.as_numpy().flatten()]

            encoded = self.tokenizer(
                texts, padding='max_length', truncation=True,
                max_length=128, return_tensors='np',
            )

            out_ids = pb_utils.Tensor('input_ids', encoded['input_ids'].astype(np.int64))
            out_mask = pb_utils.Tensor('attention_mask', encoded['attention_mask'].astype(np.int64))
            responses.append(pb_utils.InferenceResponse([out_ids, out_mask]))
        return responses

    def finalize(self):
        pass
'''

with open(f'{MODEL_REPO}/tokenizer/1/model.py', 'w') as f:
    f.write(model_py)
print(f'Wrote {MODEL_REPO}/tokenizer/1/model.py')
````

## Postprocessor Model (Python Backend)

The postprocessor takes raw logits from the ONNX model and converts them to human-readable labels and confidence scores using softmax.

### config.pbtxt

````python
os.makedirs(f'{MODEL_REPO}/postprocessor/1', exist_ok=True)

config_text = '''name: "postprocessor"
backend: "python"
max_batch_size: 8

input [
  {
    name: "logits"
    data_type: TYPE_FP32
    dims: [2]
  }
]
output [
  {
    name: "LABEL"
    data_type: TYPE_STRING
    dims: [1]
  },
  {
    name: "SCORE"
    data_type: TYPE_FP32
    dims: [1]
  }
]

instance_group [
  { count: 1, kind: KIND_CPU }
]'''

with open(f'{MODEL_REPO}/postprocessor/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/postprocessor/config.pbtxt')
````

### model.py

````python
model_py = '''import json
import numpy as np
import triton_python_backend_utils as pb_utils


class TritonPythonModel:
    def initialize(self, args):
        # Label mapping — matches DistilBERT SST-2
        self.id2label = {0: 'NEGATIVE', 1: 'POSITIVE'}

    def execute(self, requests):
        responses = []
        for request in requests:
            logits = pb_utils.get_input_tensor_by_name(request, 'logits').as_numpy()

            # Softmax
            exp_logits = np.exp(logits - np.max(logits, axis=1, keepdims=True))
            probs = exp_logits / exp_logits.sum(axis=1, keepdims=True)

            # Best prediction
            pred_indices = probs.argmax(axis=1)
            labels = np.array([[self.id2label[i]] for i in pred_indices], dtype=object)
            scores = np.array([[probs[j, pred_indices[j]]] for j in range(len(pred_indices))], dtype=np.float32)

            out_label = pb_utils.Tensor('LABEL', labels)
            out_score = pb_utils.Tensor('SCORE', scores)
            responses.append(pb_utils.InferenceResponse([out_label, out_score]))
        return responses

    def finalize(self):
        pass
'''

with open(f'{MODEL_REPO}/postprocessor/1/model.py', 'w') as f:
    f.write(model_py)
print(f'Wrote {MODEL_REPO}/postprocessor/1/model.py')
````

## Combiner Model (Python Backend)

The combiner takes predictions from **two models** and merges them into a single result. When both models agree on the label, it averages their confidence scores. When they disagree, it picks the prediction with higher confidence. An `AGREEMENT` output tells the caller whether the models agreed.

This model is used by the fan-out ensemble (below) but is defined here alongside the other building-block models.

### config.pbtxt

````python
os.makedirs(f'{MODEL_REPO}/combiner/1', exist_ok=True)

config_text = '''name: "combiner"
backend: "python"
max_batch_size: 8

input [
  { name: "LABEL_V1", data_type: TYPE_STRING, dims: [1] },
  { name: "SCORE_V1", data_type: TYPE_FP32, dims: [1] },
  { name: "LABEL_V2", data_type: TYPE_STRING, dims: [1] },
  { name: "SCORE_V2", data_type: TYPE_FP32, dims: [1] }
]
output [
  { name: "LABEL", data_type: TYPE_STRING, dims: [1] },
  { name: "SCORE", data_type: TYPE_FP32, dims: [1] },
  { name: "AGREEMENT", data_type: TYPE_STRING, dims: [1] }
]

instance_group [
  { count: 1, kind: KIND_CPU }
]'''

with open(f'{MODEL_REPO}/combiner/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/combiner/config.pbtxt')
````

### model.py

````python
model_py = '''import numpy as np
import triton_python_backend_utils as pb_utils


class TritonPythonModel:
    def initialize(self, args):
        pass

    def execute(self, requests):
        responses = []
        for request in requests:
            label_v1 = pb_utils.get_input_tensor_by_name(request, 'LABEL_V1').as_numpy()
            score_v1 = pb_utils.get_input_tensor_by_name(request, 'SCORE_V1').as_numpy()
            label_v2 = pb_utils.get_input_tensor_by_name(request, 'LABEL_V2').as_numpy()
            score_v2 = pb_utils.get_input_tensor_by_name(request, 'SCORE_V2').as_numpy()

            batch_size = len(label_v1)
            labels = []
            scores = []
            agreements = []

            for i in range(batch_size):
                l1 = label_v1[i][0].decode() if isinstance(label_v1[i][0], bytes) else str(label_v1[i][0])
                l2 = label_v2[i][0].decode() if isinstance(label_v2[i][0], bytes) else str(label_v2[i][0])
                s1 = float(score_v1[i][0])
                s2 = float(score_v2[i][0])

                agree = (l1 == l2)
                if agree:
                    # Both models agree — average their scores
                    label = l1
                    score = (s1 + s2) / 2.0
                else:
                    # Disagreement — pick the model with higher confidence
                    label = l1 if s1 >= s2 else l2
                    score = max(s1, s2)

                labels.append([label])
                scores.append([score])
                agreements.append(['AGREE' if agree else 'DISAGREE'])

            out_label = pb_utils.Tensor('LABEL', np.array(labels, dtype=object))
            out_score = pb_utils.Tensor('SCORE', np.array(scores, dtype=np.float32))
            out_agreement = pb_utils.Tensor('AGREEMENT', np.array(agreements, dtype=object))

            responses.append(pb_utils.InferenceResponse([out_label, out_score, out_agreement]))
        return responses

    def finalize(self):
        pass
'''

with open(f'{MODEL_REPO}/combiner/1/model.py', 'w') as f:
    f.write(model_py)
print(f'Wrote {MODEL_REPO}/combiner/1/model.py')
````

## Linear Ensemble (Declarative Pipeline)

An [ensemble model](https://docs.nvidia.com/deeplearning/triton-inference-server/user-guide/docs/user_guide/architecture.html#ensemble-models) chains multiple models into a pipeline using only configuration — no code, no model artifact. You define the DAG of models and how their inputs/outputs connect in `config.pbtxt`, and Triton handles the data flow.

Our linear ensemble chains: **tokenizer → distilbert_onnx → postprocessor**

### config.pbtxt

The key section is `ensemble_scheduling`, which defines the steps and how tensors flow between models:

## Fan-Out Ensemble (Parallel Multi-Model DAG)

The linear ensemble chains models in sequence. But ensembles can also express **DAGs** — directed acyclic graphs where stages fan out and fan back in. This is the most powerful thing ensembles can do that a simple chain cannot.

Our fan-out ensemble runs **both models in parallel**, postprocesses each independently, then combines the results:

```
                       ┌─────────────────┐
       text ──────────▶│    tokenizer    │
                       └────────┬────────┘
                                │
                 ┌──────────────┴──────────────┐
                 │                              │
        ┌────────▼────────┐           ┌────────▼────────┐
        │  distilbert_onnx │           │    bert_onnx    │     ◀── parallel execution
        └────────┬────────┘           └────────┬────────┘
                 │                              │
        ┌────────▼────────┐           ┌────────▼────────┐
        │  postprocessor   │           │  postprocessor   │     ◀── same model, different inputs
        └────────┬────────┘           └────────┬────────┘
                 │                              │
                 └──────────────┬──────────────┘
                                │
                       ┌────────▼────────┐
                       │    combiner     │     ◀── merges both results
                       └────────┬────────┘
                                │
                          LABEL, SCORE,
                          AGREEMENT
```

**Key insight:** Triton automatically parallelizes steps 2a/2b (both models) and steps 3a/3b (both postprocessors) because they depend only on step 1's outputs, not on each other. You get parallel execution purely from the DAG structure — no threading code needed.

The `combiner` model averages scores when both models agree on the label, and picks the higher-confidence prediction when they disagree. The `AGREEMENT` output tells you whether the models agreed.

### config.pbtxt

````python
os.makedirs(f'{MODEL_REPO}/sentiment_fanout/1', exist_ok=True)

config_text = '''name: "sentiment_fanout"
platform: "ensemble"
max_batch_size: 8

input [
  { name: "TEXT", data_type: TYPE_STRING, dims: [1] }
]
output [
  { name: "LABEL", data_type: TYPE_STRING, dims: [1] },
  { name: "SCORE", data_type: TYPE_FP32, dims: [1] },
  { name: "AGREEMENT", data_type: TYPE_STRING, dims: [1] }
]

ensemble_scheduling {
  step [
    {
      model_name: "tokenizer"
      model_version: -1
      input_map { key: "TEXT", value: "TEXT" }
      output_map { key: "input_ids", value: "tok_ids" }
      output_map { key: "attention_mask", value: "tok_mask" }
    },
    {
      model_name: "distilbert_onnx"
      model_version: -1
      input_map { key: "input_ids", value: "tok_ids" }
      input_map { key: "attention_mask", value: "tok_mask" }
      output_map { key: "logits", value: "logits_v1" }
    },
    {
      model_name: "bert_onnx"
      model_version: -1
      input_map { key: "input_ids", value: "tok_ids" }
      input_map { key: "attention_mask", value: "tok_mask" }
      output_map { key: "logits", value: "logits_v2" }
    },
    {
      model_name: "postprocessor"
      model_version: -1
      input_map { key: "logits", value: "logits_v1" }
      output_map { key: "LABEL", value: "label_v1" }
      output_map { key: "SCORE", value: "score_v1" }
    },
    {
      model_name: "postprocessor"
      model_version: -1
      input_map { key: "logits", value: "logits_v2" }
      output_map { key: "LABEL", value: "label_v2" }
      output_map { key: "SCORE", value: "score_v2" }
    },
    {
      model_name: "combiner"
      model_version: -1
      input_map { key: "LABEL_V1", value: "label_v1" }
      input_map { key: "SCORE_V1", value: "score_v1" }
      input_map { key: "LABEL_V2", value: "label_v2" }
      input_map { key: "SCORE_V2", value: "score_v2" }
      output_map { key: "LABEL", value: "LABEL" }
      output_map { key: "SCORE", value: "SCORE" }
      output_map { key: "AGREEMENT", value: "AGREEMENT" }
    }
  ]
}'''

with open(f'{MODEL_REPO}/sentiment_fanout/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/sentiment_fanout/config.pbtxt')
print()
print('Triton will automatically run distilbert_onnx and bert_onnx in parallel')
print('(they both depend only on the tokenizer output, not on each other).')
````

## Routing BLS (Imperative Pipeline)

[BLS](https://docs.nvidia.com/deeplearning/triton-inference-server/user-guide/docs/python_backend/README.html#business-logic-scripting) is a Python backend model that programmatically calls other Triton models using `pb_utils.InferenceRequest`. Unlike the declarative ensemble, BLS gives you full Python control — you can add conditional routing, error handling, logging, and dynamic model selection.

**Our routing BLS model routes requests based on average text length in the batch:**
- Average < 100 chars → entire batch goes to DistilBERT (faster, good enough for short inputs)
- Average >= 100 chars → entire batch goes to BERT (more accurate for complex inputs)

> **Note:** Routing is per-batch, not per-text — all texts in a batch go to the same model. Per-text routing would require splitting texts into separate sub-requests, which adds complexity. Per-batch routing is a practical compromise.

### config.pbtxt

````python
os.makedirs(f'{MODEL_REPO}/sentiment_bls/1', exist_ok=True)

config_text = '''name: "sentiment_bls"
backend: "python"
max_batch_size: 8

input [
  {
    name: "TEXT"
    data_type: TYPE_STRING
    dims: [1]
  }
]
output [
  {
    name: "LABEL"
    data_type: TYPE_STRING
    dims: [1]
  },
  {
    name: "SCORE"
    data_type: TYPE_FP32
    dims: [1]
  },
  {
    name: "MODEL_USED"
    data_type: TYPE_STRING
    dims: [1]
  }
]

instance_group [
  { count: 1, kind: KIND_CPU }
]'''

with open(f'{MODEL_REPO}/sentiment_bls/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/sentiment_bls/config.pbtxt')
````

### model.py

The BLS model calls the tokenizer, selects an inference model based on text length, calls the postprocessor, and returns results along with which model was used.

````python
model_py = '''import numpy as np
import triton_python_backend_utils as pb_utils


class TritonPythonModel:
    def initialize(self, args):
        pass

    def execute(self, requests):
        responses = []
        for request in requests:
            input_tensor = pb_utils.get_input_tensor_by_name(request, 'TEXT')
            texts = [t.decode('utf-8') for t in input_tensor.as_numpy().flatten()]

            # Route based on text length: short → DistilBERT, long → BERT
            avg_len = sum(len(t) for t in texts) / len(texts)
            model_name = 'distilbert_onnx' if avg_len < 100 else 'bert_onnx'

            # Step 1: Tokenize
            tok_input = pb_utils.Tensor('TEXT', input_tensor.as_numpy())
            tok_request = pb_utils.InferenceRequest(
                model_name='tokenizer', requested_output_names=['input_ids', 'attention_mask'],
                inputs=[tok_input],
            )
            tok_response = tok_request.exec()
            if tok_response.has_error():
                raise pb_utils.TritonModelException(tok_response.error().message())

            input_ids = pb_utils.get_output_tensor_by_name(tok_response, 'input_ids')
            attention_mask = pb_utils.get_output_tensor_by_name(tok_response, 'attention_mask')

            # Step 2: Inference (routed model)
            infer_request = pb_utils.InferenceRequest(
                model_name=model_name, requested_output_names=['logits'],
                inputs=[
                    pb_utils.Tensor('input_ids', input_ids.as_numpy()),
                    pb_utils.Tensor('attention_mask', attention_mask.as_numpy()),
                ],
            )
            infer_response = infer_request.exec()
            if infer_response.has_error():
                raise pb_utils.TritonModelException(infer_response.error().message())

            logits = pb_utils.get_output_tensor_by_name(infer_response, 'logits')

            # Step 3: Postprocess
            post_request = pb_utils.InferenceRequest(
                model_name='postprocessor', requested_output_names=['LABEL', 'SCORE'],
                inputs=[pb_utils.Tensor('logits', logits.as_numpy())],
            )
            post_response = post_request.exec()
            if post_response.has_error():
                raise pb_utils.TritonModelException(post_response.error().message())

            out_label = pb_utils.get_output_tensor_by_name(post_response, 'LABEL')
            out_score = pb_utils.get_output_tensor_by_name(post_response, 'SCORE')

            # Add which model was used
            model_used = np.array([[model_name]] * len(texts), dtype=object)
            out_model = pb_utils.Tensor('MODEL_USED', model_used)

            responses.append(pb_utils.InferenceResponse([
                pb_utils.Tensor('LABEL', out_label.as_numpy()),
                pb_utils.Tensor('SCORE', out_score.as_numpy()),
                out_model,
            ]))
        return responses

    def finalize(self):
        pass
'''

with open(f'{MODEL_REPO}/sentiment_bls/1/model.py', 'w') as f:
    f.write(model_py)
print(f'Wrote {MODEL_REPO}/sentiment_bls/1/model.py')
````

## Cascade BLS — Confidence-Based Escalation

The cascade pattern optimizes cost and latency by running the **fast model first** and only escalating to the **larger model** when confidence is low. This is fundamentally different from the routing BLS above — routing decides based on *input properties* (text length), while cascade decides based on *inference results* (confidence score).

```
                  ┌─────────────────┐
  text ──────────▶│    tokenizer    │
                  └────────┬────────┘
                           │
                  ┌────────▼────────┐
                  │  distilbert_onnx │◀── fast model (67M params)
                  └────────┬────────┘
                           │
                  ┌────────▼────────┐
                  │  postprocessor   │
                  └────────┬────────┘
                           │
                     confidence ≥ 0.95?
                      ╱            ╲
                   YES               NO
                    │                 │
              return result    ┌──────▼──────┐
              (DistilBERT)     │   bert_onnx  │◀── large model (110M params)
                               └──────┬──────┘
                                      │
                               ┌──────▼──────┐
                               │ postprocessor │
                               └──────┬──────┘
                                      │
                                return result
                                (BERT, escalated)
```

**Why this matters in production:** If 90% of inputs are clear-cut (high confidence), the cascade runs only the fast model for those — saving compute and latency. The 10% of ambiguous inputs get the full treatment.

### config.pbtxt

````python
os.makedirs(f'{MODEL_REPO}/cascade_bls/1', exist_ok=True)

config_text = '''name: "cascade_bls"
backend: "python"
max_batch_size: 8

input [
  { name: "TEXT", data_type: TYPE_STRING, dims: [1] }
]
output [
  { name: "LABEL", data_type: TYPE_STRING, dims: [1] },
  { name: "SCORE", data_type: TYPE_FP32, dims: [1] },
  { name: "MODEL_USED", data_type: TYPE_STRING, dims: [1] }
]

instance_group [
  { count: 1, kind: KIND_CPU }
]'''

with open(f'{MODEL_REPO}/cascade_bls/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote {MODEL_REPO}/cascade_bls/config.pbtxt')
````

### model.py

The cascade model runs DistilBERT first, checks the minimum confidence across the batch, and escalates to BERT if any sample falls below the threshold. In production, you could do per-sample escalation by splitting the batch — here we escalate the entire batch for simplicity.

````python
model_py = '''import numpy as np
import triton_python_backend_utils as pb_utils


class TritonPythonModel:
    def initialize(self, args):
        self.threshold = 0.95

    def execute(self, requests):
        responses = []
        for request in requests:
            input_tensor = pb_utils.get_input_tensor_by_name(request, 'TEXT')
            texts = [t.decode('utf-8') for t in input_tensor.as_numpy().flatten()]

            # Step 1: Tokenize
            tok_request = pb_utils.InferenceRequest(
                model_name='tokenizer',
                requested_output_names=['input_ids', 'attention_mask'],
                inputs=[pb_utils.Tensor('TEXT', input_tensor.as_numpy())],
            )
            tok_response = tok_request.exec()
            if tok_response.has_error():
                raise pb_utils.TritonModelException(tok_response.error().message())

            input_ids = pb_utils.get_output_tensor_by_name(tok_response, 'input_ids')
            attention_mask = pb_utils.get_output_tensor_by_name(tok_response, 'attention_mask')

            # Step 2: Fast model first (DistilBERT)
            fast_request = pb_utils.InferenceRequest(
                model_name='distilbert_onnx',
                requested_output_names=['logits'],
                inputs=[
                    pb_utils.Tensor('input_ids', input_ids.as_numpy()),
                    pb_utils.Tensor('attention_mask', attention_mask.as_numpy()),
                ],
            )
            fast_response = fast_request.exec()
            if fast_response.has_error():
                raise pb_utils.TritonModelException(fast_response.error().message())

            logits = pb_utils.get_output_tensor_by_name(fast_response, 'logits')

            # Step 3: Postprocess to check confidence
            post_request = pb_utils.InferenceRequest(
                model_name='postprocessor',
                requested_output_names=['LABEL', 'SCORE'],
                inputs=[pb_utils.Tensor('logits', logits.as_numpy())],
            )
            post_response = post_request.exec()
            labels = pb_utils.get_output_tensor_by_name(post_response, 'LABEL').as_numpy()
            scores = pb_utils.get_output_tensor_by_name(post_response, 'SCORE').as_numpy()

            # Step 4: Escalate if any sample is below confidence threshold
            min_confidence = float(scores.min())
            if min_confidence < self.threshold:
                # Re-run entire batch through BERT (larger, more accurate)
                slow_request = pb_utils.InferenceRequest(
                    model_name='bert_onnx',
                    requested_output_names=['logits'],
                    inputs=[
                        pb_utils.Tensor('input_ids', input_ids.as_numpy()),
                        pb_utils.Tensor('attention_mask', attention_mask.as_numpy()),
                    ],
                )
                slow_response = slow_request.exec()
                if slow_response.has_error():
                    raise pb_utils.TritonModelException(slow_response.error().message())

                post_request2 = pb_utils.InferenceRequest(
                    model_name='postprocessor',
                    requested_output_names=['LABEL', 'SCORE'],
                    inputs=[pb_utils.Tensor('logits',
                        pb_utils.get_output_tensor_by_name(slow_response, 'logits').as_numpy())],
                )
                post_response2 = post_request2.exec()
                labels = pb_utils.get_output_tensor_by_name(post_response2, 'LABEL').as_numpy()
                scores = pb_utils.get_output_tensor_by_name(post_response2, 'SCORE').as_numpy()
                model_used = 'bert_onnx (escalated)'
            else:
                model_used = 'distilbert_onnx'

            model_array = np.array([[model_used]] * len(texts), dtype=object)
            responses.append(pb_utils.InferenceResponse([
                pb_utils.Tensor('LABEL', labels),
                pb_utils.Tensor('SCORE', scores),
                pb_utils.Tensor('MODEL_USED', model_array),
            ]))
        return responses

    def finalize(self):
        pass
'''

with open(f'{MODEL_REPO}/cascade_bls/1/model.py', 'w') as f:
    f.write(model_py)
print(f'Wrote {MODEL_REPO}/cascade_bls/1/model.py')
````

## Test: Cascade BLS with Confidence-Based Escalation

Send two batches to `cascade_bls`:
- **Batch 1:** All high-confidence texts — DistilBERT handles them directly (fast path)
- **Batch 2:** Includes a borderline text ("average experience") — DistilBERT's confidence drops below 0.95, triggering escalation to BERT for the entire batch

The cascade pattern saves compute on easy inputs while ensuring accuracy on hard ones.

````python
# Batch 1: All high-confidence → DistilBERT handles it (no escalation)
HIGH_CONF = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "What a fantastic performance by the entire cast!",
]

text_array = np.array([[t] for t in HIGH_CONF], dtype=object)
inp = httpclient.InferInput('TEXT', list(text_array.shape), 'BYTES')
inp.set_data_from_numpy(text_array)

result = client.infer('cascade_bls', [inp],
    outputs=[httpclient.InferRequestedOutput('LABEL'),
             httpclient.InferRequestedOutput('SCORE'),
             httpclient.InferRequestedOutput('MODEL_USED')])

print('Batch 1 — High confidence (no escalation):\n')
for text, label, score, model in zip(HIGH_CONF,
        result.as_numpy('LABEL').flatten(),
        result.as_numpy('SCORE').flatten(),
        result.as_numpy('MODEL_USED').flatten()):
    print(f'  "{text[:55]}..."')
    print(f'    => {label.decode()} ({score:.4f}) [{model.decode()}]')
    print()

# Batch 2: Includes borderline text → triggers escalation to BERT
MIXED_CONF = [
    "This product is great!",
    "An average experience, nothing special but not bad either.",  # ~0.86 confidence
    "Absolutely terrible, worst purchase ever.",
]

text_array2 = np.array([[t] for t in MIXED_CONF], dtype=object)
inp2 = httpclient.InferInput('TEXT', list(text_array2.shape), 'BYTES')
inp2.set_data_from_numpy(text_array2)

result2 = client.infer('cascade_bls', [inp2],
    outputs=[httpclient.InferRequestedOutput('LABEL'),
             httpclient.InferRequestedOutput('SCORE'),
             httpclient.InferRequestedOutput('MODEL_USED')])

print('Batch 2 — Mixed confidence (escalation triggered):\n')
for text, label, score, model in zip(MIXED_CONF,
        result2.as_numpy('LABEL').flatten(),
        result2.as_numpy('SCORE').flatten(),
        result2.as_numpy('MODEL_USED').flatten()):
    print(f'  "{text[:55]}..."')
    print(f'    => {label.decode()} ({score:.4f}) [{model.decode()}]')
    print()
````

## Comparing All Four Patterns

| Aspect | Linear Ensemble | Fan-Out Ensemble | Routing BLS | Cascade BLS |
|--------|:---:|:---:|:---:|:---:|
| **Mechanism** | Ensemble (config) | Ensemble (config) | BLS (Python) | BLS (Python) |
| **Pipeline shape** | Fixed linear chain | Fixed DAG (parallel) | Dynamic routing | Conditional escalation |
| **Models invoked** | 1 per request | 2 in parallel | 1 (selected at runtime) | 1 or 2 (conditional) |
| **Decision logic** | None | None | Based on input properties | Based on inference results |
| **Extra output** | — | `AGREEMENT` | `MODEL_USED` | `MODEL_USED` |
| **Best for** | Simple pipelines | Model voting / ensembling | A/B testing, model tiering | Cost optimization |
| **Model artifact** | None (empty `1/` dir) | None (empty `1/` dir) | `model.py` | `model.py` |

### When to use which

- **Linear Ensemble** — Your pipeline is a straight chain with no branching: tokenize → infer → postprocess.
- **Fan-Out Ensemble** — You want to run **multiple models on the same input** and combine results (voting, averaging, consensus checking). Triton parallelizes automatically.
- **Routing BLS** — You need to **select a model based on input properties** — text length, language, category, customer tier. The decision happens before inference.
- **Cascade BLS** — You want to **optimize cost** by running the fast/cheap model first and only escalating to the slow/expensive model when the fast model is uncertain. The decision happens after inference.

## Model Swap in Ensemble

One of the ensemble's strengths is that you can swap the inference backend without changing the tokenizer or postprocessor. For example, replacing DistilBERT with BERT requires only changing one line in the ensemble config:

```protobuf
# Change this:
model_name: "distilbert_onnx"
# To this:
model_name: "bert_onnx"
```

The tokenizer works for both models because DistilBERT and BERT share the same WordPiece vocabulary (`bert-base-uncased`). The postprocessor maps index 0 → NEGATIVE and index 1 → POSITIVE, which is the correct sentiment mapping for both models (BERT's HuggingFace config uses generic `LABEL_0`/`LABEL_1` names, but index 0 is negative and index 1 is positive for both SST-2 classifiers).

Let's create a second ensemble that uses BERT and compare results.

````python
os.makedirs(f'{MODEL_REPO}/sentiment_ensemble_bert/1', exist_ok=True)

config_text = '''name: "sentiment_ensemble_bert"
platform: "ensemble"
max_batch_size: 8

input [
  { name: "TEXT", data_type: TYPE_STRING, dims: [1] }
]
output [
  { name: "LABEL", data_type: TYPE_STRING, dims: [1] },
  { name: "SCORE", data_type: TYPE_FP32, dims: [1] }
]

ensemble_scheduling {
  step [
    {
      model_name: "tokenizer"
      model_version: -1
      input_map { key: "TEXT", value: "TEXT" }
      output_map { key: "input_ids", value: "input_ids" }
      output_map { key: "attention_mask", value: "attention_mask" }
    },
    {
      model_name: "bert_onnx"
      model_version: -1
      input_map { key: "input_ids", value: "input_ids" }
      input_map { key: "attention_mask", value: "attention_mask" }
      output_map { key: "logits", value: "logits" }
    },
    {
      model_name: "postprocessor"
      model_version: -1
      input_map { key: "logits", value: "logits" }
      output_map { key: "LABEL", value: "LABEL" }
      output_map { key: "SCORE", value: "SCORE" }
    }
  ]
}'''

with open(f'{MODEL_REPO}/sentiment_ensemble_bert/config.pbtxt', 'w') as f:
    f.write(config_text)
print(f'Wrote BERT ensemble config. Restarting Triton to load it ...')

# Restart Triton to pick up the new model
subprocess.run(['docker', 'stop', CONTAINER_NAME], capture_output=True)
subprocess.run(['docker', 'rm', CONTAINER_NAME], capture_output=True)
result = subprocess.run([
    'docker', 'run', '-d', '--name', CONTAINER_NAME,
    '-p', '8000:8000', '-p', '8001:8001', '-p', '8002:8002',
    '-v', f'{os.path.abspath(MODEL_REPO)}:/models',
    TRITON_IMAGE, 'tritonserver', '--model-repository=/models', '--log-verbose=0',
], capture_output=True, text=True)
print(f'Container restarted: {result.stdout.strip()[:12]}')

# Wait for ready
client = httpclient.InferenceServerClient(url='localhost:8000')
for i in range(90):
    try:
        if client.is_server_ready():
            print(f'Triton ready ({i}s)')
            break
    except Exception:
        pass
    time.sleep(1)
````

## Performance Comparison

Compare all five inference approaches used in this notebook:
1. **ONNX direct** — client tokenizes, sends tensors, and postprocesses logits (fastest server-side, more client work)
2. **Linear Ensemble** — declarative chain: tokenizer → ONNX → postprocessor
3. **Fan-Out Ensemble** — declarative DAG: tokenizer → 2 models in parallel → 2 postprocessors → combiner
4. **Routing BLS** — imperative pipeline with text-length routing
5. **Cascade BLS** — imperative pipeline with confidence-based escalation

````python
def benchmark(fn, n=20):
    times = []
    for _ in range(n):
        start = time.time()
        fn()
        times.append((time.time() - start) * 1000)
    times.sort()
    return times[len(times) // 2]

text_array = np.array([[t] for t in TEST_INSTANCES], dtype=object)

def run_ensemble():
    inp = httpclient.InferInput('TEXT', list(text_array.shape), 'BYTES')
    inp.set_data_from_numpy(text_array)
    client.infer('sentiment_ensemble', [inp])

def run_fanout():
    inp = httpclient.InferInput('TEXT', list(text_array.shape), 'BYTES')
    inp.set_data_from_numpy(text_array)
    client.infer('sentiment_fanout', [inp])

def run_bls():
    inp = httpclient.InferInput('TEXT', list(text_array.shape), 'BYTES')
    inp.set_data_from_numpy(text_array)
    client.infer('sentiment_bls', [inp])

def run_cascade():
    inp = httpclient.InferInput('TEXT', list(text_array.shape), 'BYTES')
    inp.set_data_from_numpy(text_array)
    client.infer('cascade_bls', [inp])

# ONNX direct (for reference — client-side tokenization)
tokenizer = AutoTokenizer.from_pretrained(f'{MODELS_LOCAL}/v1')
encoded = tokenizer(TEST_INSTANCES, padding=True, truncation=True, max_length=128, return_tensors='np')

def run_onnx_direct():
    i1 = httpclient.InferInput('input_ids', list(encoded['input_ids'].shape), 'INT64')
    i1.set_data_from_numpy(encoded['input_ids'].astype(np.int64))
    i2 = httpclient.InferInput('attention_mask', list(encoded['attention_mask'].shape), 'INT64')
    i2.set_data_from_numpy(encoded['attention_mask'].astype(np.int64))
    client.infer('distilbert_onnx', [i1, i2])

# Warmup
for fn in [run_ensemble, run_fanout, run_bls, run_cascade, run_onnx_direct]:
    fn()

results = {
    'ONNX direct': benchmark(run_onnx_direct),
    'Linear Ensemble': benchmark(run_ensemble),
    'Fan-Out Ensemble': benchmark(run_fanout),
    'Routing BLS': benchmark(run_bls),
    'Cascade BLS': benchmark(run_cascade),
}

print(f'Latency comparison (median of 20 calls, batch of {len(TEST_INSTANCES)}):')
print(f'  {"Approach":<20} {"Latency (ms)":>12}  Notes')
print(f'  {"-"*20} {"-"*12}  {"-"*40}')
print(f'  {"ONNX direct":<20} {results["ONNX direct"]:>10.1f} ms  client tokenizes, no postprocess')
print(f'  {"Linear Ensemble":<20} {results["Linear Ensemble"]:>10.1f} ms  full pipeline, declarative (1 model)')
print(f'  {"Fan-Out Ensemble":<20} {results["Fan-Out Ensemble"]:>10.1f} ms  2 models parallel + combiner')
print(f'  {"Routing BLS":<20} {results["Routing BLS"]:>10.1f} ms  imperative, single model routed')
print(f'  {"Cascade BLS":<20} {results["Cascade BLS"]:>10.1f} ms  fast model + possible escalation')
print()
print(f'Linear Ensemble overhead vs ONNX direct: +{results["Linear Ensemble"] - results["ONNX direct"]:.1f} ms')
print(f'Fan-Out vs Linear Ensemble: +{results["Fan-Out Ensemble"] - results["Linear Ensemble"]:.1f} ms (runs 2x models)')
print(f'Cascade BLS vs Linear Ensemble: +{results["Cascade BLS"] - results["Linear Ensemble"]:.1f} ms (may escalate)')
````
