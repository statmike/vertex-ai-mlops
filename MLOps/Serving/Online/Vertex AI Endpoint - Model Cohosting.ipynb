# Vertex AI Endpoint — Model Cohosting

Part of [`vertex-ai-mlops/MLOps/Serving/Online`](https://github.com/statmike/vertex-ai-mlops/tree/main/MLOps/Serving/Online)

Deploy multiple models to a shared VM pool using a **DeploymentResourcePool** — a cost-efficient pattern for serving many low-traffic models from the same compute.

**What you'll learn:**
- Creating a DeploymentResourcePool with autoscaling
- Deploying multiple models to the same pool (cohosted deployment)
- Giving each model its own Endpoint URL while sharing a backend VM pool
- Scaling behavior with aggregate load across cohosted models

**Key concept:** separate endpoints, same VMs. Each model gets its own endpoint URL for independent routing and traffic management, but all models share the same underlying compute resources. This is ideal when you have many models that individually don't justify dedicated VMs.

**Important:** Co-hosting requires [Vertex AI pre-built serving containers](https://cloud.google.com/vertex-ai/docs/predictions/pre-built-containers) — custom containers are not supported. This notebook uses the PyTorch pre-built container (TorchServe) with HuggingFace models packaged as `.mar` archives.

**Models:**
- v1: [`distilbert-base-uncased-finetuned-sst-2-english`](https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english) (67M params)
- v2: [`textattack/bert-base-uncased-SST-2`](https://huggingface.co/textattack/bert-base-uncased-SST-2) (110M params)

## Package Models for Pre-built Container

DeploymentResourcePool (model co-hosting) requires [Vertex AI pre-built serving containers](https://cloud.google.com/vertex-ai/docs/predictions/pre-built-containers) — custom containers are not supported. This means we can't use the custom FastAPI container from the other notebooks in this series.

We'll use the **PyTorch pre-built container**, which runs [TorchServe](https://pytorch.org/serve/) under the hood. TorchServe expects a **Model Archive** (`.mar` file) containing:
1. The model weights and tokenizer files
2. A custom handler that defines preprocessing (tokenization), inference, and postprocessing (label mapping)

The handler runs inside the container, so tokenization happens server-side — the client sends raw text and gets back labels and scores, just like our custom FastAPI container.

```python
PYTORCH_CONTAINER = 'us-docker.pkg.dev/vertex-ai/prediction/pytorch-cpu.2-4:latest'
print(f'Pre-built container: {PYTORCH_CONTAINER}')
```

### TorchServe Handler

The handler defines how TorchServe processes each request. It extends `BaseHandler` and overrides:
- `initialize`: load model and tokenizer from the archive
- `handle`: process prediction requests end-to-end (parse input, tokenize, infer, format output)

We override `handle` directly (instead of the separate `preprocess`/`inference`/`postprocess` hooks) because Vertex AI's prediction proxy sends data in a format that differs from TorchServe's native request envelope. Overriding `handle` lets us control the full request lifecycle.

```python
%%writefile {DIR}/handler.py
import json
import inspect
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from ts.torch_handler.base_handler import BaseHandler


class SentimentHandler(BaseHandler):

    def initialize(self, context):
        properties = context.system_properties
        model_dir = properties.get("model_dir", ".")

        self.model = AutoModelForSequenceClassification.from_pretrained(model_dir)
        self.tokenizer = AutoTokenizer.from_pretrained(model_dir)
        self.model.eval()
        self.id2label = self.model.config.id2label

        # Cache the set of parameter names the model's forward() accepts.
        # DistilBERT doesn't take token_type_ids; BERT does — this lets the
        # same handler work for both without hard-coding model differences.
        self._model_params = set(
            inspect.signature(self.model.forward).parameters.keys()
        )

    def handle(self, data, context):
        """Override handle directly to bypass the request envelope.

        Vertex AI sends instances as dicts with a "body" key containing the
        JSON-serialized prediction payload ({"instances": [...]}).  TorchServe
        may also pass raw strings or bytes depending on the call path.  This
        method handles all variants.
        """
        texts = []
        for row in data:
            # Extract the raw body from whichever wrapper TorchServe uses
            if isinstance(row, dict):
                body = row.get("body") or row.get("data") or ""
            elif isinstance(row, (bytes, bytearray)):
                body = row.decode("utf-8")
            else:
                body = row

            if isinstance(body, (bytes, bytearray)):
                body = body.decode("utf-8")

            # The body may be a JSON string — e.g. '{"instances": ["text"]}'
            if isinstance(body, str):
                try:
                    parsed = json.loads(body)
                    if isinstance(parsed, dict) and "instances" in parsed:
                        texts.extend([str(t) for t in parsed["instances"]])
                        continue
                    elif isinstance(parsed, list):
                        texts.extend([str(t) for t in parsed])
                        continue
                    elif isinstance(parsed, str):
                        body = parsed
                except (json.JSONDecodeError, ValueError):
                    pass

            texts.append(str(body))

        if not texts:
            return []

        tokenized = self.tokenizer(
            texts, return_tensors="pt", padding=True, truncation=True, max_length=128,
        )

        # Only pass keys the model's forward() actually accepts
        # (e.g. DistilBERT doesn't take token_type_ids, BERT does)
        inputs = {k: v for k, v in tokenized.items() if k in self._model_params}

        with torch.no_grad():
            outputs = self.model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=1)

        results = []
        for prob in probs:
            idx = prob.argmax().item()
            results.append({
                "label": self.id2label[idx],
                "score": round(prob[idx].item(), 6),
            })
        return results
```

```python
mar_dir = f'{DIR}/model_store'
os.makedirs(mar_dir, exist_ok=True)

# Requirements file — bundled into the .mar so TorchServe installs them at model load time
requirements_path = f'{DIR}/mar_requirements.txt'
with open(requirements_path, 'w') as f:
    f.write('transformers\nsentencepiece\n')

for version, info in MODELS.items():
    gcs_mar_prefix = f'{SERIES}/{EXPERIMENT}/torchserve/{version}'

    # Check if already uploaded
    existing = list(bucket.list_blobs(prefix=f'{gcs_mar_prefix}/model.mar', max_results=1))
    if existing:
        print(f'{version} model.mar already in GCS at gs://{GCS_BUCKET}/{gcs_mar_prefix}/')
        info['mar_uri'] = f'gs://{GCS_BUCKET}/{gcs_mar_prefix}'
        continue

    print(f'Archiving {version} ({info["hub_name"]}) ...')

    # Download model + tokenizer to a temp dir
    with tempfile.TemporaryDirectory() as tmpdir:
        model = AutoModelForSequenceClassification.from_pretrained(info['hub_name'])
        tokenizer = AutoTokenizer.from_pretrained(info['hub_name'])
        model.save_pretrained(tmpdir)
        tokenizer.save_pretrained(tmpdir)

        # Collect extra-files (all model artifacts)
        extra_files = ','.join(
            os.path.join(tmpdir, f) for f in os.listdir(tmpdir) if os.path.isfile(os.path.join(tmpdir, f))
        )

        # Create .mar archive — MUST be named "model" so the file is model.mar
        # (PyTorch pre-built container requires exactly model.mar)
        subprocess.run([
            shutil.which('torch-model-archiver') or '.venv/bin/torch-model-archiver',
            '--model-name', 'model',
            '--version', '1.0',
            '--handler', f'{DIR}/handler.py',
            '--extra-files', extra_files,
            '--requirements-file', requirements_path,
            '--export-path', mar_dir,
            '--force',
        ], check=True)
        print(f'  Created {mar_dir}/model.mar')

    # Upload model.mar to a version-specific GCS prefix
    blob = bucket.blob(f'{gcs_mar_prefix}/model.mar')
    blob.upload_from_filename(f'{mar_dir}/model.mar')
    info['mar_uri'] = f'gs://{GCS_BUCKET}/{gcs_mar_prefix}'
    print(f'  Uploaded to {info["mar_uri"]}/model.mar')

# Set URIs for any that were already uploaded
for version, info in MODELS.items():
    if 'mar_uri' not in info:
        info['mar_uri'] = f'gs://{GCS_BUCKET}/{SERIES}/{EXPERIMENT}/torchserve/{version}'
```

## Create a DeploymentResourcePool

A [DeploymentResourcePool](https://cloud.google.com/vertex-ai/docs/predictions/model-co-hosting) is a shared pool of VMs that multiple deployed models share:

- **Shared VMs:** Models are deployed to the *pool*, not to individual VMs. The pool manages the underlying compute.
- **Independent Endpoints:** Each model still gets its own Endpoint for routing, traffic management, and access control.
- **Cost savings:** Instead of 5 low-traffic models each on their own `n1-standard-4`, you can cohost all 5 in a single pool — paying for 1 set of VMs instead of 5.
- **Trade-off:** Cohosted models share CPU, memory, and network. A "noisy neighbor" (one model spiking in traffic) can impact latency for other models in the pool. There is less isolation than dedicated deployments.

```python
from google.cloud.aiplatform_v1 import (
    DeploymentResourcePoolServiceClient,
    CreateDeploymentResourcePoolRequest,
    DeploymentResourcePool,
)
from google.cloud.aiplatform_v1.types import machine_resources

drp_client = DeploymentResourcePoolServiceClient(
    client_options={'api_endpoint': f'{REGION}-aiplatform.googleapis.com'},
)

POOL_ID = f'{SERIES}-{EXPERIMENT}-pool'
parent = f'projects/{PROJECT_ID}/locations/{REGION}'

pool = DeploymentResourcePool(
    dedicated_resources=machine_resources.DedicatedResources(
        machine_spec=machine_resources.MachineSpec(
            machine_type='n1-standard-4',
        ),
        min_replica_count=1,
        max_replica_count=3,
        autoscaling_metric_specs=[
            machine_resources.AutoscalingMetricSpec(
                metric_name='aiplatform.googleapis.com/prediction/online/cpu/utilization',
                target=60,
            ),
        ],
    ),
)

operation = drp_client.create_deployment_resource_pool(
    parent=parent,
    deployment_resource_pool=pool,
    deployment_resource_pool_id=POOL_ID,
)
print('Creating DeploymentResourcePool...')
pool_result = operation.result()
print(f'Pool created: {pool_result.name}')
POOL_NAME = pool_result.name
```

```python
# List existing DeploymentResourcePools in the project
print('DeploymentResourcePools in this project:')
for p in drp_client.list_deployment_resource_pools(parent=parent):
    print(f'  {p.name}')
    print(f'    machine_type: {p.dedicated_resources.machine_spec.machine_type}')
    print(f'    replicas: {p.dedicated_resources.min_replica_count}-{p.dedicated_resources.max_replica_count}')
    print()
```

### Pool Sizing

The `machine_type` and `min_replica_count` / `max_replica_count` apply to the **pool as a whole**, not per model. When you add more models to the pool, the pool's autoscaler may add more replicas (up to `max_replica_count`) to handle the aggregate load, but it will not change the machine type.

Choose your pool size based on the **total** expected load across all cohosted models, plus headroom for traffic spikes.

## Deploy Models to the Resource Pool

With cohosted deployment, each model gets its own Endpoint (for independent routing and access control) but all models share the same pool of VMs. The key API difference is using `shared_resources` (pointing to the pool) instead of `dedicated_resources` (specifying per-model machine type and replicas).

### Create Endpoints

```python
endpoint_v1 = aiplatform.Endpoint.create(
    display_name=f'{SERIES}-{EXPERIMENT}-v1',
    labels={'series': SERIES, 'experiment': EXPERIMENT.replace('-', '_'), 'version': 'v1'},
)
print(f'Endpoint v1: {endpoint_v1.resource_name}')
```

```python
endpoint_v2 = aiplatform.Endpoint.create(
    display_name=f'{SERIES}-{EXPERIMENT}-v2',
    labels={'series': SERIES, 'experiment': EXPERIMENT.replace('-', '_'), 'version': 'v2'},
)
print(f'Endpoint v2: {endpoint_v2.resource_name}')
```

### Deploy to Shared Pool

Use the gapic v1 `EndpointServiceClient` to deploy each model with `shared_resources` set to the pool name. This is the key difference from a normal deployment where you'd specify `dedicated_resources` with machine type and replica count.

```python
from google.cloud.aiplatform_v1 import (
    EndpointServiceClient,
    DeployModelRequest,
)
from google.cloud.aiplatform_v1.types import endpoint as endpoint_types

endpoint_client = EndpointServiceClient(
    client_options={'api_endpoint': f'{REGION}-aiplatform.googleapis.com'},
)

# Deploy v1 (DistilBERT) to the shared pool
deployed_model_v1 = endpoint_types.DeployedModel(
    model=vertex_model_v1.resource_name,
    display_name='distilbert-sst2',
    shared_resources=POOL_NAME,
)

operation = endpoint_client.deploy_model(
    endpoint=endpoint_v1.resource_name,
    deployed_model=deployed_model_v1,
    traffic_split={'0': 100},
)
print('Deploying DistilBERT (v1) to shared pool...')
operation.result()
print('v1 deployed to pool')
```

```python
# Deploy v2 (BERT) to the SAME shared pool
deployed_model_v2 = endpoint_types.DeployedModel(
    model=vertex_model_v2.resource_name,
    display_name='bert-sst2',
    shared_resources=POOL_NAME,
)

operation = endpoint_client.deploy_model(
    endpoint=endpoint_v2.resource_name,
    deployed_model=deployed_model_v2,
    traffic_split={'0': 100},
)
print('Deploying BERT (v2) to shared pool...')
operation.result()
print('v2 deployed to pool')
```

Both models are now deployed to the **same VMs** in the resource pool, but each has its own independent endpoint URL. Predictions sent to `endpoint_v1` always reach DistilBERT; predictions sent to `endpoint_v2` always reach BERT. The underlying compute is shared.

Send prediction requests to each endpoint and verify that they serve different models from the same pool. The TorchServe handler handles tokenization server-side, so we send raw text and get back labels and scores.

### Test Sentences

```python
TEST_INSTANCES = [
    "This movie is absolutely wonderful and I loved every minute of it!",
    "The food was terrible and the service was even worse.",
    "An average experience, nothing special but not bad either.",
]

def show_predictions(response):
    for instance, pred in zip(TEST_INSTANCES, response.predictions):
        print(f'  "{instance[:60]}..."')
        print(f'    => {pred["label"]} (score: {pred["score"]})')
        print()
```

```python
# Wait for model servers to be ready — TorchServe needs time to install
# bundled requirements (transformers, sentencepiece) and load the model
from google.api_core.exceptions import ServiceUnavailable

print('Waiting for model servers to become ready...')
for name, ep in [('v1', endpoint_v1), ('v2', endpoint_v2)]:
    for attempt in range(1, 13):
        try:
            ep.predict(instances=["test"])
            print(f'  {name}: ready')
            break
        except ServiceUnavailable:
            print(f'  {name}: not ready yet (attempt {attempt}/12), waiting 30s...')
            time.sleep(30)
    else:
        print(f'  {name}: WARNING — still not ready after 6 minutes, predictions may fail')
print('Done.')
```

```python
print('v1 (DistilBERT) predictions:\n')
response_v1 = endpoint_v1.predict(instances=TEST_INSTANCES)
show_predictions(response_v1)
```

```python
print('v2 (BERT) predictions:\n')
response_v2 = endpoint_v2.predict(instances=TEST_INSTANCES)
show_predictions(response_v2)
```

The different label formats confirm that different models are serving from the same pool:
- **DistilBERT** returns `POSITIVE`/`NEGATIVE` labels (human-readable names from its config)
- **BERT** returns `LABEL_0`/`LABEL_1` labels (numeric IDs from the textattack config)

### Scaling Behavior

The pool scales based on **aggregate** load across all cohosted models. If DistilBERT is idle but BERT is under heavy load, the pool scales up — and DistilBERT benefits from the extra capacity even though it didn't trigger the scaling.

This is both a benefit (free capacity from neighbors' scaling) and a risk (your model's scaling is tied to neighbors' load patterns).

### Cohosted vs. Separate Deployments

| Aspect | Cohosted (Resource Pool) | Separate Deployments |
|--------|-------------------------|---------------------|
| VMs | Shared pool | Dedicated per model |
| Cost | Lower (shared compute) | Higher (dedicated compute) |
| Isolation | Lower (noisy neighbor risk) | Full isolation |
| Scaling | Aggregate load | Per-model |
| Endpoints | One per model | One per model |
| Best for | Many low-traffic models | High-traffic, SLA-critical |

---
## When to Cohost

### DO cohost when:
- You have **many models with low individual traffic** — each doesn't justify its own VM
- You're running **dev/staging environments** where isolation is less critical
- You need **cost-sensitive deployments** and can tolerate shared resources
- Your models have **complementary traffic patterns** (e.g., one peaks mornings, another peaks evenings)

### DON'T cohost when:
- Models have **different SLAs** — a low-priority model could degrade a high-priority one
- You're deploying **GPU-heavy models** — GPU memory contention is harder to manage than CPU
- Models have **unpredictable traffic spikes** — one model's burst can starve others
- You need **production-critical isolation** — dedicated deployments provide full resource guarantees
