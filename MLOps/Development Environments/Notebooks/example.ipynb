# Bayesian Change Point Detection with PyMC

A marketing team notices their daily website signups look different lately — but *when* did the change happen, and *how big* is it?

This notebook uses [PyMC](https://www.pymc.io/) to build a Bayesian change point model that answers both questions simultaneously, with full uncertainty quantification.

## What We'll Build

1. **Generate** synthetic daily signup data with a hidden change point
2. **Model** the data as two Poisson processes separated by an unknown switchpoint
3. **Infer** the posterior distributions for the change point and both rates
4. **Visualize** the results with ArviZ and matplotlib

Change point detection is a classic Bayesian problem — the model doesn't just give a single answer, it gives a *distribution* over possible answers, showing how confident it is.

```python
import subprocess, sys, shutil

def install(*packages):
    """Install packages using uv (fast) with pip fallback."""
    uv = shutil.which('uv')
    if uv:
        subprocess.check_call([uv, 'pip', 'install', '-q', '--python', sys.executable, *packages])
    else:
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', '--upgrade', *packages])

install('pymc', 'arviz', 'matplotlib', 'numpy')
```

```python
import pymc as pm
import arviz as az
import numpy as np
import matplotlib.pyplot as plt

print(f'PyMC version: {pm.__version__}')
print(f'ArviZ version: {az.__version__}')
```

---
## Generate Synthetic Data

We simulate 200 days of website signups. The true (hidden) parameters:
- **Days 0–119:** ~15 signups/day (Poisson distributed)
- **Days 120–199:** ~25 signups/day (something changed!)

The model's job is to recover these parameters from the data alone.

```python
np.random.seed(42)

n_days = 200
true_switchpoint = 120
true_early_rate = 15
true_late_rate = 25

early_data = np.random.poisson(true_early_rate, true_switchpoint)
late_data = np.random.poisson(true_late_rate, n_days - true_switchpoint)
signups = np.concatenate([early_data, late_data])

print(f'Generated {n_days} days of signup data')
print(f'  Mean (first half):  {signups[:100].mean():.1f}')
print(f'  Mean (second half): {signups[100:].mean():.1f}')
```

---
## Visualize the Raw Data

Can you spot when the change happened just by looking?

```python
fig, ax = plt.subplots(figsize=(12, 4))

ax.scatter(range(n_days), signups, s=15, alpha=0.6, color='#5E81AC', edgecolors='none')
ax.set_xlabel('Day', fontsize=12)
ax.set_ylabel('Daily Signups', fontsize=12)
ax.set_title('Daily Website Signups — When Did Something Change?', fontsize=14, fontweight='bold')

# Add a rolling average to hint at the pattern
window = 14
rolling_avg = np.convolve(signups, np.ones(window)/window, mode='valid')
ax.plot(range(window-1, n_days), rolling_avg, color='#BF616A', linewidth=2, label=f'{window}-day rolling avg')
ax.legend(fontsize=11)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

---
## Build the Bayesian Model

The model has three unknown parameters:

| Parameter | Prior | Interpretation |
|---|---|---|
| `switchpoint` | DiscreteUniform(0, 199) | The day the rate changed |
| `early_rate` | Exponential(1/20) | Average signups/day before the change |
| `late_rate` | Exponential(1/20) | Average signups/day after the change |

The observed signups follow a Poisson distribution whose rate switches at the (unknown) switchpoint.

```python
with pm.Model() as changepoint_model:
    # Priors
    switchpoint = pm.DiscreteUniform('switchpoint', lower=0, upper=n_days - 1)
    early_rate = pm.Exponential('early_rate', lam=1.0 / 20)
    late_rate = pm.Exponential('late_rate', lam=1.0 / 20)

    # The rate switches at the switchpoint
    rate = pm.math.switch(switchpoint >= np.arange(n_days), early_rate, late_rate)

    # Observed data
    observations = pm.Poisson('signups', mu=rate, observed=signups)

print(changepoint_model)
```

---
## Sample from the Posterior

PyMC uses MCMC (Markov Chain Monte Carlo) to draw samples from the posterior distribution. We run 4 chains for robust diagnostics.

```python
with changepoint_model:
    trace = pm.sample(2000, tune=1000, chains=4, return_inferencedata=True, random_seed=42)

print(f'Sampling complete: {trace.posterior.dims}')
```

---
## Posterior Summary

The summary table shows the posterior mean, standard deviation, and 94% HDI (Highest Density Interval) for each parameter.

```python
az.summary(trace, var_names=['switchpoint', 'early_rate', 'late_rate'], round_to=2)
```

---
## Trace Plots

Trace plots show the sampling history (right) and the posterior distribution (left) for each parameter. Good mixing looks like a "fuzzy caterpillar" in the trace.

```python
az.plot_trace(
    trace,
    var_names=['switchpoint', 'early_rate', 'late_rate'],
    figsize=(12, 7),
    compact=True,
)
plt.suptitle('MCMC Trace Plots', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

---
## Posterior Distributions

Each posterior shows the model's belief about a parameter after seeing the data. The shaded region is the 94% HDI — the most credible range of values.

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

colors = ['#5E81AC', '#A3BE8C', '#BF616A']
var_names = ['switchpoint', 'early_rate', 'late_rate']
true_values = [true_switchpoint, true_early_rate, true_late_rate]
labels = ['Change Point (day)', 'Early Rate (signups/day)', 'Late Rate (signups/day)']

for ax, var, true_val, color, label in zip(axes, var_names, true_values, colors, labels):
    az.plot_posterior(
        trace, var_names=[var], ax=ax, color=color,
        hdi_prob=0.94, point_estimate='mean',
    )
    ax.axvline(true_val, color='#2E3440', linestyle='--', linewidth=2, label=f'True = {true_val}')
    ax.set_title(label, fontsize=12, fontweight='bold')
    ax.legend(fontsize=10)

plt.suptitle('Posterior Distributions vs. True Values', fontsize=14, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
```

---
## The Big Picture — Model Meets Data

The final plot overlays the model's inferences on the original data:
- **Blue shaded region** — 94% HDI for the change point location
- **Dashed lines** — inferred early and late rates
- The model should recover parameters close to the true values

```python
fig, ax = plt.subplots(figsize=(14, 5))

# Raw data
ax.scatter(range(n_days), signups, s=15, alpha=0.4, color='#4C566A', edgecolors='none', label='Observed signups')

# Posterior means
sp_mean = float(trace.posterior['switchpoint'].mean())
er_mean = float(trace.posterior['early_rate'].mean())
lr_mean = float(trace.posterior['late_rate'].mean())

# Change point HDI
sp_hdi = az.hdi(trace, var_names=['switchpoint'], hdi_prob=0.94)['switchpoint'].values
ax.axvspan(sp_hdi[0], sp_hdi[1], alpha=0.15, color='#5E81AC', label=f'Switchpoint 94% HDI [{sp_hdi[0]:.0f}, {sp_hdi[1]:.0f}]')
ax.axvline(sp_mean, color='#5E81AC', linewidth=2.5, linestyle='-', label=f'Switchpoint mean = {sp_mean:.0f}')

# True change point
ax.axvline(true_switchpoint, color='#2E3440', linewidth=2, linestyle=':', alpha=0.7, label=f'True switchpoint = {true_switchpoint}')

# Rate lines
ax.hlines(er_mean, 0, sp_mean, colors='#A3BE8C', linewidth=3, linestyle='-', label=f'Early rate = {er_mean:.1f} (true: {true_early_rate})')
ax.hlines(lr_mean, sp_mean, n_days, colors='#BF616A', linewidth=3, linestyle='-', label=f'Late rate = {lr_mean:.1f} (true: {true_late_rate})')

ax.set_xlabel('Day', fontsize=12)
ax.set_ylabel('Daily Signups', fontsize=12)
ax.set_title('Bayesian Change Point Detection — Model vs. Reality', fontsize=14, fontweight='bold')
ax.legend(loc='upper left', fontsize=10, framealpha=0.9)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print(f'\nModel recovered:')
print(f'  Switchpoint: {sp_mean:.0f} (true: {true_switchpoint})')
print(f'  Early rate:  {er_mean:.1f} (true: {true_early_rate})')
print(f'  Late rate:   {lr_mean:.1f} (true: {true_late_rate})')
```

---
## Summary

The Bayesian change point model successfully identified:
- **When** the pattern shifted — the posterior for `switchpoint` is tightly concentrated near the true value
- **How much** it shifted — both `early_rate` and `late_rate` posteriors bracket the true rates
- **How confident** we should be — the HDI intervals give honest uncertainty bounds

This is the power of Bayesian modeling: instead of a single point estimate, you get a full picture of what the data supports.
