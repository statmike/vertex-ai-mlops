```python
graph_attr = dict(bgcolor = 'white', labelloc = 'topCenter', fontcolor = 'black')
node_attr = dict(shape = 'rectangle', style = 'filled,rounded', fontcolor = gcolors['blue'][0], fillcolor = gcolors['blue'][3])
subgraph_attr = dict(style = 'filled,rounded', labelloc = 'topCenter', fontcolor = gcolors['blue'][0], fillcolor = gcolors['blue'][8])
sub_subgraph_attr = dict(style = 'filled,rounded', labelloc = 'topCenter', fontcolor = gcolors['blue'][0], fillcolor = gcolors['blue'][4])
edge_attr = dict(style = '')
```

```python
# given level of current row:
# check for previous row at current level, then climb up to l1 looking for prior row
# at each level of climb, also dig down to nested levels looking for prior row

def edge_digger(end_node, structure):
    # called by edger function only
    # iterate down the levels of the hierarchy lookin gfor a previous row to the current one
    if type(structure) == str:
        g.edge(structure, end_node, **edge_attr)
    else:
        for c in structure:
            for r in c[-1:]:
                if type(r) == str:
                    g.edge(r, end_node, **edge_attr)
                else:
                    edge_digger(end_node, c)
    return

def edger(path, end_node, structure):
    # path = (row l1, col l1, row l2, col l2, ...), end_node = 'node_l1_...', structure (so far)
    
    # iterate up the levels of the hierarchy looking for a previous row to the current one.  i = 1 will use the current level, i = 2 the next hier, ...
    ri = 1
    while (2*(len(path) // 2) - (2 * ri)) >= 0:
        row_index = (2*(len(path) // 2) - (2 * ri)) # 4 for l3, 2 for l2, 0 for l1
        if path[row_index] > 0: # is the row at this level not the first row?
            for rs in range(row_index + 1):
                i = path[rs]
                if rs == row_index: i -= 1
                if rs == 0:
                    sub_structure = structure[i]
                else:
                    sub_structure = sub_structure[i]
            edge_digger(end_node, sub_structure)
            break;
        else: # the row at current level is the first row, 0, move up to higher level
            ri += 1

    return
```

```python
def add_node(name, label, path, structure):
    g.node(name = name, label = label, **node_attr)
    
    if len(path) // 2 == 1:
        if row > 0: edger(path, name, structure)
    else: edger(path, name, structure)
    
    return



def add_level(vertex_arch, level, path, name, structure):
    shape = vertex_arch['shape']
    parts = vertex_arch['parts']
    
    with g.subgraph(name = f'cluster_l1_{l1_iter}') as s:
        s.attr(label = l1_part['name'], **subgraph_attr)
    
        for row in range(shape[0]):
            structure.append([])
            for col in range(shape[1]):
                structure[row].append([])
                part = parts[level_iter]
                path += (row,col)

                # determine next step:
                if :
                    # add subgraph
                    add_level()
                else:
                    # add node
                    add_node()
                level += 1
        
    
    return





# settings
graph_levels = ['l1', 'l2', 'l3', 'l4']

# BEGIN GRAPH
g = graphviz.Digraph()
g.attr(label = vertex_arch['name'], **graph_attr)

structure = []
path = ()
name = 'node_'
add_level(vertex_arch['l1'], 1, path, name, structure)







level_iter = 0
for row in range(shape[0]):
    structure.append([])
    for col in range(shape[1]):
        structure[row].append([])
        part = parts[level_iter]
        path += (row, col)
        if part[f'l{level + 1}'] and f'l{level + 1}' in graph levels:
            add_subgraph()
        else:
            structure[row][col].append(name + f'l{level + 1}_{level_iter}')
            add_node(structure[row][col][-1], part['name'], path, structure)
        level_iter += 1
```
