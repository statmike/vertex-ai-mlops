# local icons for dot/graphviz to use

```python
import requests
```

```python
def write_file(svg_url):
    svg = requests.get(svg_url).text
    with open(svg_url.split('/')[-3]+'.svg', 'w') as file:
        file.write(svg)
```

```python
url = 'https://fonts.gstatic.com/s/i/gcpiconscolors/automl_tables/v1/32px.svg'
```

```python
write_file(url)
```
