# QRCode On Demand

Develop a process that creates a QRCode as a response.

```python
#!pip install qrcode[pil] -U -q --user
```

```python
import qrcode
```

```python
img = qrcode.make('https://www.google.com')
img
```

```python
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=4,
    border=1,
)
qr.add_data('Some data')
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color="white")
img
```

```python
img = qr.make_image(back_color=(0, 0, 256), fill_color=(256, 0, 0))
img
```

```python
import qrcode.image.styledpil
import qrcode.image.styles.moduledrawers.pil
import qrcode.image.styles.colormasks

img = qr.make_image(
    image_factory = qrcode.image.styledpil.StyledPilImage,
    module_drawer = qrcode.image.styles.moduledrawers.pil.GappedSquareModuleDrawer(),
    color_mask = qrcode.image.styles.colormasks.SquareGradiantColorMask()
)
img
```
