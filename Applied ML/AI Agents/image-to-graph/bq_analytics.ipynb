# BigQuery Agent Analytics

This notebook provides queries for analyzing `agent_image_to_graph` behavior using the [BigQuery Agent Analytics Plugin](https://google.github.io/adk-docs/observability/bigquery-agent-analytics/).

**No manual setup required** — the agent automatically creates the BigQuery dataset and table on first run via `agent_image_to_graph/bq_plugin.py`. This notebook is purely for querying and analyzing the logged events.

Analytics are configured in `.env`:
```
BQ_ANALYTICS_DATASET=applied_ml_image_to_graph
BQ_ANALYTICS_TABLE=agent_events
BQ_ANALYTICS_GCS_BUCKET=your-bucket-name
BQ_ANALYTICS_GCS_PATH=applied-ml/ai-agents/image-to-graph/bq_plugin
```

## Connect

````python
import os
from google.cloud import bigquery
from dotenv import load_dotenv

load_dotenv()

PROJECT_ID = os.getenv('GOOGLE_CLOUD_PROJECT')
DATASET_ID = os.getenv('BQ_ANALYTICS_DATASET', 'applied_ml_image_to_graph')
TABLE_ID = os.getenv('BQ_ANALYTICS_TABLE', 'agent_events')
FULL_TABLE_ID = f"{PROJECT_ID}.{DATASET_ID}.{TABLE_ID}"

bq_client = bigquery.Client(project=PROJECT_ID)

# Verify the table exists (auto-created by the agent on first run)
try:
    table = bq_client.get_table(FULL_TABLE_ID)
    print(f"Connected to: {FULL_TABLE_ID}")
    print(f"  Rows: {table.num_rows:,}")
    print(f"  Size: {table.num_bytes / 1024 / 1024:.2f} MB")
    print(f"  Created: {table.created}")
except Exception as e:
    print(f"Table not found: {FULL_TABLE_ID}")
    print(f"Run the agent first — it auto-creates the dataset and table on startup.")
````

## Bounding Box Corrections

Track how often bounding box auto-correction fires during graph construction. Two correction types are tracked via `bbox_corrections` in `STATE_DELTA` events:

1. **Per-node inversions** (`update_graph`): `normalize_bbox()` fixes individual nodes with inverted axes (min > max) or out-of-range values
2. **Axis swap** (`validate_graph`): Detects and corrects systematic x/y coordinate confusion across all nodes using dual-signal detection (element aspect ratios + group extreme ratios)

````python
import json as _json
import pandas as pd

bbox_df = bq_client.query(f"""
WITH corrections AS (
    SELECT
        session_id,
        JSON_QUERY(attributes, '$.state_delta.bbox_corrections') AS corrections_json
    FROM `{FULL_TABLE_ID}`
    WHERE event_type = 'STATE_DELTA'
      AND JSON_QUERY(attributes, '$.state_delta.bbox_corrections') IS NOT NULL
)
SELECT session_id, corrections_json
FROM corrections
""").to_dataframe()

if not bbox_df.empty:
    # Each STATE_DELTA captures the full accumulated list, so take the last
    # (largest) entry per session to avoid double-counting
    latest = bbox_df.sort_values('corrections_json', key=lambda s: s.str.len()).groupby('session_id').tail(1)

    # Parse and separate correction types
    node_rows = []
    swap_rows = []
    for _, row in latest.iterrows():
        sid = row['session_id'][-8:]
        entries = _json.loads(row['corrections_json'])
        for e in entries:
            if e.get('op') == 'xy_swap':
                swap_rows.append({
                    'session_id': sid,
                    'nodes_swapped': e.get('node_count', 0),
                    'image_size': f"{e.get('image_width', '?')}x{e.get('image_height', '?')}",
                })
            elif e.get('node_id'):
                node_rows.append({
                    'session_id': sid,
                    'node_id': e.get('node_id'),
                    'op': e.get('op'),
                    'y_inverted': e.get('y_inverted', False),
                    'x_inverted': e.get('x_inverted', False),
                    'clamped': e.get('clamped', False),
                })

    # --- Axis swap summary ---
    if swap_rows:
        swap_df = pd.DataFrame(swap_rows)
        print("Axis Swap Corrections (validate_graph)")
        print("=" * 50)
        display(swap_df)
        print()
    else:
        print("No axis swap corrections needed.\n")

    # --- Per-node inversion summary ---
    if node_rows:
        corr_df = pd.DataFrame(node_rows)
        summary = corr_df.groupby('session_id').agg(
            corrected_nodes=('node_id', 'count'),
            y_inverted=('y_inverted', 'sum'),
            x_inverted=('x_inverted', 'sum'),
            clamped=('clamped', 'sum'),
        ).reset_index()

        print("Per-Node Bbox Inversions (update_graph)")
        print("=" * 50)
        display(summary)

        # Bar chart: corrections per session
        ax = summary.plot.bar(
            x='session_id',
            y=['y_inverted', 'x_inverted', 'clamped'],
            stacked=True,
            title='Per-Node Bbox Corrections by Type',
            ylabel='Count',
            xlabel='Session',
            rot=45,
        )
        ax.figure.tight_layout()
    else:
        print("No per-node bbox inversions needed.")

else:
    print("No bbox correction data found in STATE_DELTA events.")
````

# Usage Analysis

This agent uses two models with different tracking mechanisms:

| Model Role | Model | How Calls Are Made | How Usage Is Tracked |
|---|---|---|---|
| **Agent (orchestration)** | `AGENT_MODEL` (e.g. gemini-3-flash-preview) | ADK orchestration pipeline | `LLM_RESPONSE` events — captured automatically by the BQ Analytics Plugin |
| **Tool (vision/analysis)** | `TOOL_MODEL` (e.g. gemini-3.1-pro-preview) | Direct `google.genai` calls in `util_gemini.generate_content()` | `STATE_DELTA` events — usage stored in `tool_context.state["tool_llm_usage"]` during tool execution |

The tool-level calls (Pro model for image analysis) are the most expensive. They are tracked by writing per-call usage data into `tool_context.state` during tool execution. The BQ Analytics Plugin automatically captures state changes as `STATE_DELTA` events after each tool completes.

## Discover Event Schema

Inspect the actual JSON structure logged by the ADK BigQuery Analytics Plugin to confirm field paths for token counts and model name.

````python
import json as _json

# Inspect LLM_RESPONSE: content and attributes
print("=== LLM_RESPONSE ===")
resp_df = bq_client.query(f"""
SELECT content, attributes
FROM `{FULL_TABLE_ID}`
WHERE event_type = 'LLM_RESPONSE'
LIMIT 1
""").to_dataframe()

if len(resp_df) > 0:
    resp_content = _json.loads(resp_df.iloc[0]['content'])
    print(f"content keys: {list(resp_content.keys())}")
    if 'usage' in resp_content:
        print(f"  usage: {resp_content['usage']}")
    resp_attrs = resp_df.iloc[0]['attributes']
    if resp_attrs:
        attrs = _json.loads(resp_attrs)
        print(f"attributes keys: {list(attrs.keys())}")
        if 'model_version' in attrs:
            print(f"  model_version: {attrs['model_version']}")
        if 'usage_metadata' in attrs:
            print(f"  usage_metadata: {attrs['usage_metadata']}")
    else:
        print("attributes: None")

# Inspect LLM_REQUEST: content and attributes
print("\n=== LLM_REQUEST ===")
req_df = bq_client.query(f"""
SELECT content, attributes
FROM `{FULL_TABLE_ID}`
WHERE event_type = 'LLM_REQUEST'
LIMIT 1
""").to_dataframe()

if len(req_df) > 0:
    req_content = _json.loads(req_df.iloc[0]['content'])
    print(f"content keys: {list(req_content.keys())}")
    req_attrs = req_df.iloc[0]['attributes']
    if req_attrs:
        attrs = _json.loads(req_attrs)
        print(f"attributes keys: {list(attrs.keys())}")
        if 'model' in attrs:
            print(f"  model: {attrs['model']}")
    else:
        print("attributes: None")
````

## Per-Session Usage

Each session is one agent run — one image evaluated into one graph. This is the unit of cost. The query below combines both **agent LLM** calls (`LLM_RESPONSE`) and **tool LLM** calls (`STATE_DELTA`) per session, broken down by model.

Input and output tokens are tracked separately since they have different pricing. Note that **total tokens may exceed input + output** — the difference is thinking/reasoning tokens (visible in Gemini's `usage_metadata.thoughts_token_count`).

````python
import pandas as pd

# Combined per-session per-model usage: agent LLM (LLM_RESPONSE) + tool LLM (STATE_DELTA)
session_model_df = bq_client.query(f"""
WITH agent_calls AS (
    SELECT
        session_id,
        COALESCE(
            JSON_VALUE(attributes, '$.model_version'),
            JSON_VALUE(attributes, '$.model'),
            'unknown'
        ) as model,
        'agent' as source,
        CAST(JSON_VALUE(content, '$.usage.prompt') AS INT64) as prompt_tokens,
        CAST(JSON_VALUE(content, '$.usage.completion') AS INT64) as completion_tokens,
        CAST(JSON_VALUE(content, '$.usage.total') AS INT64) as total_tokens
    FROM `{FULL_TABLE_ID}`
    WHERE event_type = 'LLM_RESPONSE'
),
tool_deltas AS (
    SELECT
        session_id,
        invocation_id,
        JSON_QUERY(attributes, '$.state_delta.tool_llm_usage') as usage_json
    FROM `{FULL_TABLE_ID}`
    WHERE event_type = 'STATE_DELTA'
      AND JSON_QUERY(attributes, '$.state_delta.tool_llm_usage') IS NOT NULL
),
tool_latest AS (
    SELECT
        session_id,
        invocation_id,
        ARRAY_AGG(usage_json ORDER BY ARRAY_LENGTH(JSON_QUERY_ARRAY(usage_json)) DESC LIMIT 1)[OFFSET(0)] as usage_json
    FROM tool_deltas
    GROUP BY session_id, invocation_id
),
tool_calls AS (
    SELECT
        session_id,
        JSON_VALUE(entry, '$.model') as model,
        'tool' as source,
        CAST(JSON_VALUE(entry, '$.prompt_tokens') AS INT64) as prompt_tokens,
        CAST(JSON_VALUE(entry, '$.completion_tokens') AS INT64) as completion_tokens,
        CAST(JSON_VALUE(entry, '$.total_tokens') AS INT64) as total_tokens
    FROM tool_latest,
    UNNEST(JSON_QUERY_ARRAY(usage_json)) as entry
),
all_calls AS (
    SELECT * FROM agent_calls
    UNION ALL
    SELECT * FROM tool_calls
)
SELECT
    session_id,
    model,
    source,
    COUNT(*) as calls,
    SUM(prompt_tokens) as prompt_tokens,
    SUM(completion_tokens) as completion_tokens,
    SUM(total_tokens) as total_tokens
FROM all_calls
GROUP BY session_id, model, source
ORDER BY session_id, model, source
""").to_dataframe()

if len(session_model_df) > 0:
    # Pivot: one row per session, columns for each model's tokens
    pivot = session_model_df.pivot_table(
        index='session_id',
        columns=['model', 'source'],
        values=['calls', 'prompt_tokens', 'completion_tokens', 'total_tokens'],
        aggfunc='sum',
        fill_value=0,
    )
    pivot.columns = ['_'.join(col).strip() for col in pivot.columns]
    pivot = pivot.reset_index()

    # Compute session totals by source and token type
    def _sum_cols(df, prefix, contains=None):
        cols = [c for c in df.columns if c.startswith(prefix)]
        if contains:
            cols = [c for c in cols if contains in c]
        return df[cols].sum(axis=1) if cols else 0

    pivot['agent_input'] = _sum_cols(pivot, 'prompt_tokens_', '_agent')
    pivot['agent_output'] = _sum_cols(pivot, 'completion_tokens_', '_agent')
    pivot['agent_total'] = _sum_cols(pivot, 'total_tokens_', '_agent')
    pivot['tool_input'] = _sum_cols(pivot, 'prompt_tokens_', '_tool')
    pivot['tool_output'] = _sum_cols(pivot, 'completion_tokens_', '_tool')
    pivot['tool_total'] = _sum_cols(pivot, 'total_tokens_', '_tool')
    pivot['session_input'] = pivot['agent_input'] + pivot['tool_input']
    pivot['session_output'] = pivot['agent_output'] + pivot['tool_output']
    pivot['session_total'] = pivot['agent_total'] + pivot['tool_total']
    pivot = pivot.sort_values('session_total', ascending=False)

    # Display concise view: per-model totals + session summaries
    all_token_cols = sorted([c for c in pivot.columns if c.startswith('total_tokens_')])
    display_cols = (['session_id'] + all_token_cols
                    + ['agent_input', 'agent_output', 'agent_total',
                       'tool_input', 'tool_output', 'tool_total',
                       'session_input', 'session_output', 'session_total'])
    display(pivot[display_cols])

    # Per-session summary
    print()
    for _, row in pivot.iterrows():
        sid = row['session_id'][-8:]
        print(f"  ...{sid}  agent: {int(row['agent_input']):,} in + {int(row['agent_output']):,} out"
              f" | tool: {int(row['tool_input']):,} in + {int(row['tool_output']):,} out"
              f" | session: {int(row['session_input']):,} in + {int(row['session_output']):,} out"
              f" = {int(row['session_total']):,}")
else:
    print("No usage data found. Run the agent to generate data.")
````

## Tool LLM Usage (STATE_DELTA)

Tool-level Gemini calls (via `util_gemini.generate_content()`) store per-call usage in `tool_context.state["tool_llm_usage"]`. The BQ Analytics Plugin captures these as `STATE_DELTA` events after each tool completes.

Each entry in the `tool_llm_usage` array contains:
- `tool` — which tool made the call (e.g. `analyze_image`, `crop_and_examine`)
- `model` — the model used (e.g. `gemini-3.1-pro-preview`)
- `prompt_tokens`, `completion_tokens`, `total_tokens` — token counts for the call

### Discover STATE_DELTA Format

Inspect the actual JSON structure of `STATE_DELTA` events that contain `tool_llm_usage` to confirm field paths.

### Session Statistics

Summary statistics across sessions for cost estimation. With more sessions, these averages become more reliable for predicting costs of future batches.

````python
# Across-session statistics: avg, min, max, median tokens per session per model
if len(session_model_df) > 0:
    # Per-session totals by source, split by input/output
    session_source = session_model_df.groupby(['session_id', 'source']).agg(
        input_tokens=('prompt_tokens', 'sum'),
        output_tokens=('completion_tokens', 'sum'),
        total_tokens=('total_tokens', 'sum'),
    ).reset_index()

    # Overall session totals
    session_overall = session_model_df.groupby('session_id').agg(
        input_tokens=('prompt_tokens', 'sum'),
        output_tokens=('completion_tokens', 'sum'),
        total_tokens=('total_tokens', 'sum'),
    ).reset_index()

    # Per-model session totals
    session_by_model = session_model_df.groupby(['session_id', 'model']).agg(
        input_tokens=('prompt_tokens', 'sum'),
        output_tokens=('completion_tokens', 'sum'),
        total_tokens=('total_tokens', 'sum'),
    ).reset_index()

    print(f"Sessions: {session_overall['session_id'].nunique()}\n")

    # Per-model statistics
    print("Per-model statistics (tokens per session):")
    for model in sorted(session_by_model['model'].unique()):
        md = session_by_model[session_by_model['model'] == model]
        print(f"  {model}:")
        for label, col in [('input', 'input_tokens'), ('output', 'output_tokens'), ('total', 'total_tokens')]:
            print(f"    {label:6s}:  avg {md[col].mean():>10,.0f}"
                  f" | median {md[col].median():>10,.0f}"
                  f" | min {md[col].min():>10,}"
                  f" | max {md[col].max():>10,}")

    # Overall statistics
    print(f"\nOverall (tokens per session):")
    for label, col in [('input', 'input_tokens'), ('output', 'output_tokens'), ('total', 'total_tokens')]:
        print(f"  {label:6s}:  avg {session_overall[col].mean():>10,.0f}"
              f" | median {session_overall[col].median():>10,.0f}"
              f" | min {session_overall[col].min():>10,}"
              f" | max {session_overall[col].max():>10,}")

    # Plain-language summary
    agent_src = session_source[session_source['source'] == 'agent']
    tool_src = session_source[session_source['source'] == 'tool']
    print(f"\nAvg session: ~{session_overall['total_tokens'].mean()/1000:.0f}K tokens total"
          f" (~{agent_src['total_tokens'].mean()/1000:.0f}K agent"
          f" + ~{tool_src['total_tokens'].mean()/1000:.0f}K tool)")
    print(f"  input:  ~{session_overall['input_tokens'].mean()/1000:.0f}K"
          f" (~{agent_src['input_tokens'].mean()/1000:.0f}K agent"
          f" + ~{tool_src['input_tokens'].mean()/1000:.0f}K tool)")
    print(f"  output: ~{session_overall['output_tokens'].mean()/1000:.0f}K"
          f" (~{agent_src['output_tokens'].mean()/1000:.0f}K agent"
          f" + ~{tool_src['output_tokens'].mean()/1000:.0f}K tool)")
else:
    print("No usage data found. Run the agent to generate data.")
````

### Tool LLM Usage By Tool

Same data grouped by tool name — shows per-tool token consumption and call counts.

````python
# Tool LLM usage by tool name
tool_llm_tool_df = bq_client.query(f"""
WITH deltas AS (
    SELECT
        session_id,
        invocation_id,
        JSON_QUERY(attributes, '$.state_delta.tool_llm_usage') as usage_json
    FROM `{FULL_TABLE_ID}`
    WHERE event_type = 'STATE_DELTA'
      AND JSON_QUERY(attributes, '$.state_delta.tool_llm_usage') IS NOT NULL
),
latest AS (
    SELECT
        session_id,
        invocation_id,
        ARRAY_AGG(usage_json ORDER BY ARRAY_LENGTH(JSON_QUERY_ARRAY(usage_json)) DESC LIMIT 1)[OFFSET(0)] as usage_json
    FROM deltas
    GROUP BY session_id, invocation_id
),
entries AS (
    SELECT
        session_id,
        JSON_VALUE(entry, '$.tool') as tool,
        JSON_VALUE(entry, '$.model') as model,
        CAST(JSON_VALUE(entry, '$.prompt_tokens') AS INT64) as prompt_tokens,
        CAST(JSON_VALUE(entry, '$.completion_tokens') AS INT64) as completion_tokens,
        CAST(JSON_VALUE(entry, '$.total_tokens') AS INT64) as total_tokens
    FROM latest,
    UNNEST(JSON_QUERY_ARRAY(usage_json)) as entry
)
SELECT
    tool,
    COUNT(*) as calls,
    SUM(prompt_tokens) as prompt_tokens,
    SUM(completion_tokens) as completion_tokens,
    SUM(total_tokens) as total_tokens,
    ROUND(AVG(prompt_tokens)) as avg_prompt,
    ROUND(AVG(completion_tokens)) as avg_completion
FROM entries
GROUP BY tool
ORDER BY total_tokens DESC
""").to_dataframe()

if len(tool_llm_tool_df) > 0:
    display(tool_llm_tool_df)
else:
    print("No tool LLM usage data found. Run the agent to generate data.")
````

````python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# 1. Per-session stacked bars by model+source (input vs output)
ax1 = axes[0]
if len(session_model_df) > 0:
    sm = session_model_df.groupby(['session_id', 'model', 'source']).agg(
        input_tokens=('prompt_tokens', 'sum'),
        output_tokens=('completion_tokens', 'sum'),
    ).reset_index()
    # Create separate input and output segments per model+source
    segments = []
    for _, row in sm.drop_duplicates(['model', 'source'])[['model', 'source']].iterrows():
        label_base = f"{row['model']} ({row['source']})"
        segments.append((f"{label_base} input", 'input_tokens', row['model'], row['source']))
        segments.append((f"{label_base} output", 'output_tokens', row['model'], row['source']))

    # Build pivot with input/output as separate columns
    sm['label_in'] = sm['model'] + ' (' + sm['source'] + ') input'
    sm['label_out'] = sm['model'] + ' (' + sm['source'] + ') output'
    in_pivot = sm.pivot_table(index='session_id', columns='label_in',
                              values='input_tokens', fill_value=0)
    out_pivot = sm.pivot_table(index='session_id', columns='label_out',
                               values='output_tokens', fill_value=0)
    combined = pd.concat([in_pivot, out_pivot], axis=1)
    combined = combined.loc[combined.sum(axis=1).sort_values(ascending=False).index]

    session_labels = ['...' + s[-8:] for s in combined.index]
    bottom = [0] * len(combined)
    # Alternate light/dark pairs: input lighter, output darker
    color_pairs = [('#7baaf7', '#1a73e8'), ('#f28b82', '#c5221f'),
                   ('#fdd663', '#e8a900'), ('#81c995', '#1e8e3e')]
    col_colors = {}
    pair_idx = 0
    seen_bases = {}
    for col in sorted(combined.columns):
        base = col.rsplit(' ', 1)[0]  # strip " input" / " output"
        if base not in seen_bases:
            seen_bases[base] = pair_idx
            pair_idx += 1
        pi = seen_bases[base]
        is_output = col.endswith(' output')
        col_colors[col] = color_pairs[pi % len(color_pairs)][1 if is_output else 0]

    for col in sorted(combined.columns):
        vals = combined[col].tolist()
        ax1.barh(session_labels, vals, left=bottom,
                 label=col, color=col_colors[col])
        bottom = [b + v for b, v in zip(bottom, vals)]
    ax1.set_xlabel('Tokens')
    ax1.set_title('Tokens Per Session (input/output by model)')
    ax1.legend(loc='lower right', fontsize=6)
    ax1.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))

# 2. Input vs output tokens by model (grouped bar)
ax2 = axes[1]
if len(session_model_df) > 0:
    model_io = session_model_df.groupby('model').agg(
        input_tokens=('prompt_tokens', 'sum'),
        output_tokens=('completion_tokens', 'sum'),
    ).reset_index().sort_values('input_tokens', ascending=True)
    y_pos = np.arange(len(model_io))
    bar_h = 0.35
    ax2.barh(y_pos - bar_h/2, model_io['input_tokens'], bar_h,
             label='Input', color='#1a73e8')
    ax2.barh(y_pos + bar_h/2, model_io['output_tokens'], bar_h,
             label='Output', color='#ea4335')
    ax2.set_yticks(y_pos)
    ax2.set_yticklabels(model_io['model'])
    ax2.set_xlabel('Tokens')
    ax2.set_title('Input vs Output Tokens by Model')
    ax2.legend()
    ax2.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))

# 3. Tool LLM by tool name (input vs output)
ax3 = axes[2]
if len(tool_llm_tool_df) > 0:
    y_pos = np.arange(len(tool_llm_tool_df))
    bar_h = 0.35
    ax3.barh(y_pos - bar_h/2, tool_llm_tool_df['prompt_tokens'], bar_h,
             label='Input', color='#1a73e8')
    ax3.barh(y_pos + bar_h/2, tool_llm_tool_df['completion_tokens'], bar_h,
             label='Output', color='#ea4335')
    ax3.set_yticks(y_pos)
    ax3.set_yticklabels(tool_llm_tool_df['tool'])
    ax3.set_xlabel('Tokens')
    ax3.set_title('Tool LLM Tokens By Tool')
    ax3.legend()
    ax3.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))
else:
    ax3.text(0.5, 0.5, 'No tool LLM data', ha='center', va='center',
             transform=ax3.transAxes)

plt.tight_layout()
plt.show()
````
