# BigQuery Agent Analytics

This notebook provides queries for analyzing `agent_image_to_graph` behavior using the [BigQuery Agent Analytics Plugin](https://google.github.io/adk-docs/observability/bigquery-agent-analytics/).

**No manual setup required** — the agent automatically creates the BigQuery dataset and table on first run via `agent_image_to_graph/bq_plugin.py`. This notebook is purely for querying and analyzing the logged events.

Analytics are configured in `.env`:
```
BQ_ANALYTICS_DATASET=applied_ml_image_to_graph
BQ_ANALYTICS_TABLE=agent_events
BQ_ANALYTICS_GCS_BUCKET=your-bucket-name
BQ_ANALYTICS_GCS_PATH=applied-ml/ai-agents/image-to-graph/bq_plugin
```

## Connect

````python
import os
from google.cloud import bigquery
from dotenv import load_dotenv

load_dotenv()

PROJECT_ID = os.getenv('GOOGLE_CLOUD_PROJECT')
DATASET_ID = os.getenv('BQ_ANALYTICS_DATASET', 'applied_ml_image_to_graph')
TABLE_ID = os.getenv('BQ_ANALYTICS_TABLE', 'agent_events')
FULL_TABLE_ID = f"{PROJECT_ID}.{DATASET_ID}.{TABLE_ID}"

bq_client = bigquery.Client(project=PROJECT_ID)

# Verify the table exists (auto-created by the agent on first run)
try:
    table = bq_client.get_table(FULL_TABLE_ID)
    print(f"Connected to: {FULL_TABLE_ID}")
    print(f"  Rows: {table.num_rows:,}")
    print(f"  Size: {table.num_bytes / 1024 / 1024:.2f} MB")
    print(f"  Created: {table.created}")
except Exception as e:
    print(f"Table not found: {FULL_TABLE_ID}")
    print(f"Run the agent first — it auto-creates the dataset and table on startup.")
````

## Optional: Truncate Table

**Warning:** Deletes all data. Use only for demo resets.

````python
TRUNCATE_TABLE = False

if TRUNCATE_TABLE:
    confirm = input(f"DELETE ALL DATA from '{FULL_TABLE_ID}'? Type 'yes': ")
    if confirm.lower() == 'yes':
        bq_client.query(f"TRUNCATE TABLE `{FULL_TABLE_ID}`").result()
        print(f"Table truncated.")
    else:
        print("Cancelled.")
else:
    print("Skipping (TRUNCATE_TABLE = False)")
````

# Analytics Queries

## Recent Events

````python
bq_client.query(f"""
SELECT timestamp, event_type, agent, session_id, status
FROM `{FULL_TABLE_ID}`
ORDER BY timestamp DESC
LIMIT 20
""").to_dataframe()
````

## Event Type Distribution

````python
bq_client.query(f"""
SELECT event_type, COUNT(*) as count
FROM `{FULL_TABLE_ID}`
GROUP BY event_type
ORDER BY count DESC
""").to_dataframe()
````

## Agent Activity

````python
bq_client.query(f"""
SELECT 
    agent,
    COUNT(*) as events,
    COUNT(DISTINCT session_id) as sessions,
    AVG(CAST(JSON_VALUE(latency_ms, '$.total_ms') AS INT64)) as avg_latency_ms
FROM `{FULL_TABLE_ID}`
WHERE event_type = 'LLM_RESPONSE'
GROUP BY agent
ORDER BY events DESC
""").to_dataframe()
````

## Token Usage

````python
bq_client.query(f"""
SELECT
    agent,
    COUNT(*) as responses,
    AVG(CAST(JSON_VALUE(content, '$.usage.total') AS INT64)) as avg_tokens,
    SUM(CAST(JSON_VALUE(content, '$.usage.total') AS INT64)) as total_tokens
FROM `{FULL_TABLE_ID}`
WHERE event_type = 'LLM_RESPONSE'
GROUP BY agent
ORDER BY total_tokens DESC
""").to_dataframe()
````

## Latency Analysis

````python
bq_client.query(f"""
SELECT
    event_type,
    COUNT(*) as count,
    AVG(CAST(JSON_VALUE(latency_ms, '$.total_ms') AS INT64)) as avg_ms,
    MIN(CAST(JSON_VALUE(latency_ms, '$.total_ms') AS INT64)) as min_ms,
    MAX(CAST(JSON_VALUE(latency_ms, '$.total_ms') AS INT64)) as max_ms
FROM `{FULL_TABLE_ID}`
WHERE event_type IN ('LLM_RESPONSE', 'TOOL_COMPLETED')
GROUP BY event_type
""").to_dataframe()
````

## Tool Usage

````python
bq_client.query(f"""
SELECT
    JSON_VALUE(content, '$.tool') as tool_name,
    COUNT(*) as calls,
    AVG(CAST(JSON_VALUE(latency_ms, '$.total_ms') AS INT64)) as avg_latency_ms
FROM `{FULL_TABLE_ID}`
WHERE event_type = 'TOOL_COMPLETED'
GROUP BY tool_name
ORDER BY calls DESC
""").to_dataframe()
````

## Errors

````python
df = bq_client.query(f"""
SELECT timestamp, agent, event_type, error_message, session_id
FROM `{FULL_TABLE_ID}`
WHERE status = 'ERROR' OR error_message IS NOT NULL
ORDER BY timestamp DESC
LIMIT 20
""").to_dataframe()

print(f"Errors: {len(df)}" if len(df) > 0 else "No errors found.")
df
````

## Conversation Tracing

Trace a specific conversation by `invocation_id` or `trace_id`.

````python
# List recent invocations
bq_client.query(f"""
SELECT DISTINCT 
    invocation_id,
    trace_id,
    MIN(timestamp) as started,
    COUNT(*) as events
FROM `{FULL_TABLE_ID}`
GROUP BY invocation_id, trace_id
ORDER BY started DESC
LIMIT 10
""").to_dataframe()
````

````python
# Set an ID to trace (copy from above)
TRACE_ID = None       # e.g., "e-4999e6e4-..."
INVOCATION_ID = None  # Alternative if trace_id is NULL

id_column = "trace_id" if TRACE_ID else "invocation_id" if INVOCATION_ID else None
id_value = TRACE_ID or INVOCATION_ID

if id_value:
    df = bq_client.query(f"""
    SELECT 
        timestamp, event_type, agent,
        COALESCE(JSON_VALUE(content, '$.tool'), '') as tool,
        CAST(JSON_VALUE(latency_ms, '$.total_ms') AS INT64) as latency_ms,
        status
    FROM `{FULL_TABLE_ID}`
    WHERE {id_column} = '{id_value}'
    ORDER BY timestamp ASC
    """).to_dataframe()
    print(f"Trace: {len(df)} events")
    display(df)
else:
    print("Set TRACE_ID or INVOCATION_ID above.")
````

## Span Hierarchy

````python
if id_value:
    bq_client.query(f"""
    SELECT
        span_id, parent_span_id, event_type, timestamp,
        CAST(JSON_VALUE(latency_ms, '$.total_ms') AS INT64) as duration_ms,
        COALESCE(JSON_VALUE(content, '$.tool'), 'LLM_CALL') as operation
    FROM `{FULL_TABLE_ID}`
    WHERE {id_column} = '{id_value}'
      AND event_type IN ('LLM_RESPONSE', 'TOOL_COMPLETED')
    ORDER BY timestamp ASC
    """).to_dataframe()
else:
    print("Set TRACE_ID or INVOCATION_ID above.")
````

## Multimodal Content (GCS References)

````python
df = bq_client.query(f"""
SELECT
    timestamp, event_type,
    part.mime_type,
    part.storage_mode,
    part.object_ref.uri as gcs_uri
FROM `{FULL_TABLE_ID}`,
UNNEST(content_parts) as part
WHERE part.storage_mode = 'GCS_REFERENCE'
ORDER BY timestamp DESC
LIMIT 20
""").to_dataframe()

print(f"GCS references: {len(df)}" if len(df) > 0 else "No multimodal content in GCS.")
df
````

## Daily Activity

````python
bq_client.query(f"""
SELECT
    DATE(timestamp) as date,
    agent,
    COUNT(*) as events,
    COUNT(DISTINCT session_id) as sessions,
    COUNTIF(event_type = 'LLM_RESPONSE') as llm_calls,
    COUNTIF(event_type = 'TOOL_COMPLETED') as tool_calls
FROM `{FULL_TABLE_ID}`
GROUP BY date, agent
ORDER BY date DESC, events DESC
LIMIT 30
""").to_dataframe()
````

# Cost Analysis

## Inspect LLM_RESPONSE Content Structure

Before building cost queries, inspect the actual JSON structure of a logged `LLM_RESPONSE` event to discover the exact field paths for token counts and model name.

````python
# Inspect the content JSON structure from a sample LLM_RESPONSE event
import json as _json

sample_df = bq_client.query(f"""
SELECT content
FROM `{FULL_TABLE_ID}`
WHERE event_type = 'LLM_RESPONSE'
LIMIT 1
""").to_dataframe()

if len(sample_df) > 0:
    content = _json.loads(sample_df.iloc[0]['content'])
    print(_json.dumps(content, indent=2)[:3000])
    print("\n--- Top-level keys ---")
    print(list(content.keys()))
    if 'usage' in content:
        print("\n--- usage keys ---")
        print(content['usage'])
else:
    print("No LLM_RESPONSE events found. Run the agent first.")
````

## Pricing Configuration

Define per-model token pricing (per 1M tokens). Update these rates as pricing changes.

> **Note:** The field paths below (`content.usage.input_tokens`, `content.usage.output_tokens`, `content.model`) match the ADK BigQuery Analytics Plugin schema. If your logged events use different paths, update the SQL queries accordingly based on the inspection above.

````python
# Pricing per 1M tokens (USD) — update as needed
# Source: https://cloud.google.com/vertex-ai/generative-ai/pricing
MODEL_PRICING = {
    'gemini-2.5-flash': {'input': 0.15, 'output': 0.60},
    'gemini-2.5-pro':   {'input': 1.25, 'output': 10.00},
    'gemini-2.0-flash': {'input': 0.10, 'output': 0.40},
    'gemini-3.0-flash': {'input': 0.10, 'output': 0.40},
    'gemini-3.1-pro':   {'input': 1.25, 'output': 10.00},
}

# Build SQL CASE expression for input/output cost per token
def _pricing_case(token_type: str) -> str:
    """Generate SQL CASE for cost-per-token lookup by model name pattern."""
    lines = ["CASE"]
    for pattern, rates in MODEL_PRICING.items():
        rate = rates[token_type]
        lines.append(f"    WHEN model LIKE '%{pattern}%' THEN {rate} / 1000000.0")
    # Default fallback (assume cheapest flash pricing)
    fallback = 0.10 / 1000000.0 if token_type == 'input' else 0.40 / 1000000.0
    lines.append(f"    ELSE {fallback}")
    lines.append("  END")
    return "\n".join(lines)

print("Pricing configured for models:", list(MODEL_PRICING.keys()))
print(f"\nExample SQL (input cost/token):\n{_pricing_case('input')}")
````

## Cost Per Session

````python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Query cost breakdown by session and agent
cost_query = f"""
WITH token_costs AS (
    SELECT
        session_id,
        agent,
        JSON_VALUE(content, '$.model') as model,
        CAST(JSON_VALUE(content, '$.usage.input_tokens') AS INT64) as input_tokens,
        CAST(JSON_VALUE(content, '$.usage.output_tokens') AS INT64) as output_tokens
    FROM `{FULL_TABLE_ID}`
    WHERE event_type = 'LLM_RESPONSE'
      AND JSON_VALUE(content, '$.usage.input_tokens') IS NOT NULL
)
SELECT
    session_id,
    agent,
    model,
    SUM(input_tokens) as total_input_tokens,
    SUM(output_tokens) as total_output_tokens,
    SUM(input_tokens * {_pricing_case('input')}) as input_cost,
    SUM(output_tokens * {_pricing_case('output')}) as output_cost,
    SUM(input_tokens * {_pricing_case('input')} + output_tokens * {_pricing_case('output')}) as total_cost
FROM token_costs
GROUP BY session_id, agent, model
ORDER BY session_id, agent
"""

cost_df = bq_client.query(cost_query).to_dataframe()

if len(cost_df) > 0:
    print(f"Sessions: {cost_df['session_id'].nunique()}")
    print(f"Total cost: ${cost_df['total_cost'].sum():.4f}")
    display(cost_df)
else:
    print("No LLM_RESPONSE events with token data found.")
````

````python
if len(cost_df) > 0:
    # Aggregate for charts
    session_cost = cost_df.groupby('session_id').agg(
        input_cost=('input_cost', 'sum'),
        output_cost=('output_cost', 'sum'),
        total_cost=('total_cost', 'sum'),
    ).reset_index()
    session_cost['session_label'] = [
        s[:8] + '...' if len(s) > 12 else s for s in session_cost['session_id']
    ]

    agent_cost = cost_df.groupby('agent')['total_cost'].sum()

    fig, axes = plt.subplots(1, 3, figsize=(16, 5))

    # 1. Bar chart: total cost per session
    ax1 = axes[0]
    ax1.bar(session_cost['session_label'], session_cost['total_cost'], color='#1a73e8')
    ax1.set_title('Cost Per Session')
    ax1.set_ylabel('Cost (USD)')
    ax1.yaxis.set_major_formatter(mticker.FormatStrFormatter('$%.4f'))
    ax1.tick_params(axis='x', rotation=45)

    # 2. Stacked bar: input vs output cost per session
    ax2 = axes[1]
    ax2.bar(session_cost['session_label'], session_cost['input_cost'],
            label='Input tokens', color='#34a853')
    ax2.bar(session_cost['session_label'], session_cost['output_cost'],
            bottom=session_cost['input_cost'], label='Output tokens', color='#ea4335')
    ax2.set_title('Input vs Output Cost')
    ax2.set_ylabel('Cost (USD)')
    ax2.yaxis.set_major_formatter(mticker.FormatStrFormatter('$%.4f'))
    ax2.legend()
    ax2.tick_params(axis='x', rotation=45)

    # 3. Pie chart: cost by agent
    ax3 = axes[2]
    if len(agent_cost) > 0:
        colors = ['#1a73e8', '#f9ab00', '#34a853', '#ea4335'][:len(agent_cost)]
        ax3.pie(agent_cost.values, labels=agent_cost.index, autopct='%1.1f%%',
                colors=colors, startangle=90)
        ax3.set_title('Cost by Agent')
    else:
        ax3.text(0.5, 0.5, 'No data', ha='center', va='center')
        ax3.set_title('Cost by Agent')

    plt.tight_layout()
    plt.show()
else:
    print("No cost data to chart.")
````
