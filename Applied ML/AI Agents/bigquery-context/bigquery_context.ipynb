# BigQuery Context — Five Approaches to Table Discovery

This notebook walks through the first three of five different approaches to finding the best BigQuery tables for answering a user's question. All approaches:

1. Discover candidate tables using different metadata sources
2. Feed the metadata to a shared **Gemini-powered reranker** that produces structured output
3. Return a `RerankerResponse` with ranked tables, confidence scores, column hints, and SQL suggestions

By running the approaches against the same question, you can compare:
- Which tables each approach finds
- How much metadata each approach provides
- How confidence scores and SQL hints differ
- The trade-offs between completeness, speed, and metadata richness

**As ADK Agents**: All five approaches are packaged as [ADK](https://google.github.io/adk-docs/) agents that run in parallel via a `ParallelAgent`. See the [readme](readme.md) for details on running them with `uv run adk web .` (interactive).

## Approach 1: BQ Metadata Tools

This approach uses the BigQuery API to enumerate datasets and tables, then inspects schemas to gather metadata. It's the most straightforward method — systematically walk through each dataset, list tables, and read their schemas.

**ADK Agent:** `agent_bq_tools` — uses the ADK `BigQueryToolset` (filtered to `list_dataset_ids`, `list_table_ids`, `get_dataset_info`, `get_table_info`). The agent makes **per-query API calls** — each user question triggers a fresh enumeration of datasets and tables to gather metadata, which is then passed to the shared reranker.

### Step 1: List Datasets in Scope

````python
datasets = get_datasets()
print(f"Datasets in scope ({len(datasets)}):")
for ds in datasets:
    dataset = bq_client.get_dataset(f"{GOOGLE_CLOUD_PROJECT}.{ds}")
    print(f"  {ds}: {dataset.description}")
````

### Step 2: List Tables in Each Dataset

````python
all_tables = {}
for ds in datasets:
    scoped = get_scoped_tables(ds)
    if scoped is None:
        # All tables — discover via BQ API
        tables = list(bq_client.list_tables(f"{GOOGLE_CLOUD_PROJECT}.{ds}"))
        all_tables[ds] = [t.table_id for t in tables]
        print(f"  {ds}: {all_tables[ds]} (all)")
    else:
        # Specific tables from scope
        all_tables[ds] = scoped
        print(f"  {ds}: {all_tables[ds]} (scoped)")

print(f"\nTotal tables: {sum(len(v) for v in all_tables.values())}")
````

### Step 3: Get Table Details

For each table, retrieve the full schema with column names, types, and descriptions.

````python
bq_metadata_parts = []
bq_nominated_ids = []

for ds, table_ids in all_tables.items():
    for table_id in table_ids:
        full_id = f"{GOOGLE_CLOUD_PROJECT}.{ds}.{table_id}"
        table = bq_client.get_table(full_id)
        bq_nominated_ids.append(full_id)

        part = f"## {full_id}\n"
        part += f"Description: {table.description or '(none)'}\n"
        part += f"Rows: {table.num_rows:,}\n"
        part += f"Columns:\n"
        for field in table.schema:
            desc = f" - {field.description}" if field.description else ""
            part += f"  - {field.name} ({field.field_type}){desc}\n"

        bq_metadata_parts.append(part)

        # Show what's being gathered
        print(f"  {full_id}")
        print(f"    Description: {(table.description or '(none)')[:80]}")
        print(f"    Columns ({len(table.schema)}): {', '.join(f.name for f in table.schema[:5])}", end="")
        if len(table.schema) > 5:
            print(f" ... +{len(table.schema) - 5} more")
        else:
            print()
        cols_with_desc = sum(1 for f in table.schema if f.description)
        print(f"    Column descriptions: {cols_with_desc}/{len(table.schema)}")
        print(f"    Metadata size: {len(part):,} chars")
        print()

bq_candidate_metadata = "\n".join(bq_metadata_parts)
print(f"Total metadata for reranker: {len(bq_candidate_metadata):,} chars across {len(bq_metadata_parts)} tables")
print(f"Nominated {len(bq_nominated_ids)} table(s)")
````

### Step 4: Rerank with Gemini Structured Output

Pass all gathered metadata to the reranker, which uses Gemini to produce a `RerankerResponse` with ranked tables, confidence scores, column hints, and SQL suggestions.

````python
result_bq_tools = call_reranker(
    question=QUESTION,
    candidate_metadata=bq_candidate_metadata,
    discovery_method="bq_tools",
    top_k=TOP_K,
)

print(f"Ranked {len(result_bq_tools.ranked_tables)} tables:\n")
for t in result_bq_tools.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    print()
````

````python
print(result_bq_tools.model_dump_json(indent=2))
````

## Approach 2: Dataplex Search

This approach uses Dataplex's semantic search to find relevant tables using natural language. It sends the user's question to `search_entries` with `semantic_search=True`, which uses embeddings to find tables whose names, descriptions, and metadata relate to the query. The search is scoped to our configured datasets using the `parent:` query qualifier so that results come from in-scope datasets rather than the entire project. Each match is then looked up for full schema details.

**ADK Agent:** `agent_dataplex_search` — the `before_agent_callback` runs semantic search, entry lookup, and reranking. The agent makes **per-query API calls** — each user question triggers a new semantic search and entry lookups, with results passed to the shared reranker.

### Step 1: Semantic Search via Dataplex

````python
catalog_client = dataplex_v1.CatalogServiceClient()

# Semantic search: scope to our datasets using the parent: qualifier.
# This tells Dataplex to only return entries from our scoped datasets,
# so most of the 20 results will be relevant instead of being pruned.
dataset_filter = "|".join(f"datasets/{ds}" for ds in get_datasets())
query = f"({QUESTION}) AND system=BIGQUERY AND parent:({dataset_filter})"
print(f"Search query:\n  {query}\n")

request = dataplex_v1.SearchEntriesRequest(
    name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/global",
    query=query,
    page_size=20,
    semantic_search=True,
)

search_results = []
skipped = 0
for result in catalog_client.search_entries(request=request):
    entry = result.dataplex_entry
    source = entry.entry_source
    fqn = entry.fully_qualified_name or ""

    # Post-filter for table-level scope (parent: only filters by dataset)
    parts = fqn.rsplit(".", 2)
    if len(parts) >= 2:
        ds_name, tbl_name = parts[-2], parts[-1]
        if not is_table_in_scope(ds_name, tbl_name):
            skipped += 1
            continue

    search_results.append({
        "entry_name": entry.name,
        "fully_qualified_name": fqn,
        "display_name": source.display_name if source else "",
        "description": source.description if source else "",
    })
    print(f"  Found: {source.display_name if source else entry.name}")

print(f"\nIn-scope results: {len(search_results)} (filtered out {skipped})")
````

### Step 2: Lookup Entry Details

For each search result, retrieve full metadata including schema, storage info, and any data profile aspects.

````python
catalog_metadata_parts = []
search_nominated_ids = []

for sr in search_results:
    request = dataplex_v1.LookupEntryRequest(
        name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/{BQ_LOCATION.lower()}",
        entry=sr["entry_name"],
        view=dataplex_v1.EntryView.FULL,
    )
    entry = catalog_client.lookup_entry(request=request)
    entry_dict = json_format.MessageToDict(entry._pb)

    # Extract key metadata
    source = entry_dict.get("entrySource", {})
    aspects = entry_dict.get("aspects", {})

    # Parse fully_qualified_name to project.dataset.table format
    # FQN format: bigquery:project.dataset.table
    fqn = sr.get("fully_qualified_name", "")
    fqn_clean = fqn.split(":", 1)[-1] if ":" in fqn else fqn
    table_label = fqn_clean or sr["display_name"]
    search_nominated_ids.append(table_label)

    part = f"## {table_label}\n"
    part += f"Description: {source.get('description', '(none)')}\n"

    # Schema
    for key, aspect in aspects.items():
        if "schema" in key:
            schema_data = aspect.get("data", {})
            fields = schema_data.get("fields", [])
            part += f"Columns ({len(fields)}):\n"
            for fld in fields:
                desc = fld.get("description", "")
                part += f"  - {fld.get('name', '?')} ({fld.get('type', '?')}) {desc}\n"

    catalog_metadata_parts.append(part)
    print(f"  {table_label}: {len(part):,} chars of metadata")

catalog_candidate_metadata = "\n".join(catalog_metadata_parts)
print(f"\nTotal metadata: {len(catalog_candidate_metadata):,} chars")
print(f"Nominated {len(search_nominated_ids)} table(s)")
````

### Step 3: Rerank

````python
if not catalog_candidate_metadata.strip():
    print("No in-scope tables found by semantic search — skipping reranker.")
    print("This can happen if Dataplex hasn't indexed the tables yet.")
    print("Try re-running after a few minutes.")
    result_dataplex_search = RerankerResponse(question=QUESTION, top_k=TOP_K, ranked_tables=[], notes="No in-scope results from semantic search.")
else:
    result_dataplex_search = call_reranker(
        question=QUESTION,
        candidate_metadata=catalog_candidate_metadata,
        discovery_method="dataplex_search",
        top_k=TOP_K,
    )

    print(f"Ranked {len(result_dataplex_search.ranked_tables)} tables:\n")
    for t in result_dataplex_search.ranked_tables:
        print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
        print(f"     {t.reasoning}")
        print()
````

#### Full Reranker Output

````python
print(result_dataplex_search.model_dump_json(indent=2))
````

## Approach 3: Dataplex Context

This approach uses the Dataplex `lookupContext` API to fetch pre-formatted, LLM-ready metadata capsules for all tables at once. The capsules include schema, column descriptions, and — when data profiling has been run — column-level statistics like null ratios, distinct values, and sample values.

> **Note:** The `lookupContext` API is not yet in the `google-cloud-dataplex` Python SDK (v2.16.0). This notebook calls it via REST using `google.auth` credentials.

**ADK Agent:** `agent_dataplex_context` — uses the `initialize_context` tool which calls `lookupContext` and caches the result in agent state. Unlike the other two approaches, metadata is fetched **once at startup** (zero per-query cost). When a user question arrives, the agent passes the pre-cached context directly to the shared reranker — no additional API calls are made. This makes it the fastest approach at query time, and the richest in metadata when Dataplex profiling has been run.

### Step 1: Fetch Context Capsules

Fetch both dataset-level context (descriptions) and table-level context (schema + data profiles) in batched API calls.

````python
# Note: importing from context_cache also populates the shared context cache
# (used by approaches 4 and 5). The manual API calls below demonstrate how
# lookupContext works step-by-step.
from context_cache.util_lookup_context import lookup_context, lookup_context_batched

# Build table entry names based on scope, tracking project.dataset.table labels
datasets = get_datasets()
table_entries = []  # list of (full_bq_id, entry_name)
for ds in datasets:
    scoped = get_scoped_tables(ds)
    if scoped is None:
        # All tables — discover via BQ API
        for t in bq_client.list_tables(f"{GOOGLE_CLOUD_PROJECT}.{ds}"):
            full_id = f"{GOOGLE_CLOUD_PROJECT}.{ds}.{t.table_id}"
            table_entries.append((full_id, get_dataplex_entry_name(ds, t.table_id)))
    else:
        # Specific tables — use directly
        for table_name in scoped:
            full_id = f"{GOOGLE_CLOUD_PROJECT}.{ds}.{table_name}"
            table_entries.append((full_id, get_dataplex_entry_name(ds, table_name)))
print(f"Resolved {len(table_entries)} tables across {len(datasets)} datasets")

# Dataset-level context
dataset_entry_names = [get_dataplex_dataset_entry_name(ds) for ds in datasets]
print(f"\nFetching context for {len(dataset_entry_names)} datasets...")
dataset_context = lookup_context_batched(dataset_entry_names)
print(f"  Dataset context: {len(dataset_context):,} chars")

# Table-level context with project.dataset.table headers
print(f"Fetching context for {len(table_entries)} tables...")
table_context_parts = []
for full_id, entry_name in table_entries:
    ctx = lookup_context([entry_name])
    if ctx:
        table_context_parts.append(f"## {full_id}\n{ctx}")
table_context = "\n\n".join(table_context_parts)
print(f"  Table context: {len(table_context):,} chars")

# Combined
dataplex_context = f"# Dataset Context\n{dataset_context}\n\n# Table Context\n{table_context}"
print(f"\nTotal knowledge context: {len(dataplex_context):,} chars")
````

#### Preview: Dataplex Context Capsule

This is what the `lookupContext` API returns — LLM-ready metadata in YAML format. Notice how data profiling adds column-level statistics (null ratios, distinct values, sample values).

````python
# Show first 3000 chars of the knowledge context
preview = dataplex_context[:3000]
print(preview)
if len(dataplex_context) > 3000:
    print(f"\n... ({len(dataplex_context) - 3000:,} more chars)")
````

### Step 2: Rerank

````python
result_dataplex_context = call_reranker(
    question=QUESTION,
    candidate_metadata=dataplex_context,
    discovery_method="dataplex_context",
    top_k=TOP_K,
)

print(f"Ranked {len(result_dataplex_context.ranked_tables)} tables:\n")
for t in result_dataplex_context.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    print()
````

````python
print(result_dataplex_context.model_dump_json(indent=2))
````

## Approach 4: Context Pre-Filter

This approach uses the shared context cache to show the LLM **brief summaries** of all in-scope tables. The LLM reviews the briefs and **nominates candidates** — tables it believes are relevant to the question. Only those nominated tables get their full detailed context sent to the reranker.

This is a hybrid approach: LLM reasoning selects candidates, then deterministic reranking scores them. The LLM sees briefs (compact: description + column names/types + row count), while the reranker sees detailed YAML capsules (schema + data profiles).

**ADK Agent:** `agent_context_prefilter` — the LLM calls a `nominate_tables` tool to store its selections in state, then an `after_agent_callback` fetches detailed cached context for the nominees and calls the shared reranker. Zero per-query discovery API calls — everything comes from the shared cache.

### Step 1: Brief Summaries from Cache

The shared context cache provides compact **brief** summaries for every in-scope table — description, column names/types, and row count. This is what the LLM reviews to decide which tables are worth detailed analysis.

````python
from context_cache import get_all_briefs, get_detailed_for_tables, get_table_ids

all_briefs = get_all_briefs()
print(f"Brief summaries for {len(get_table_ids())} tables ({len(all_briefs):,} chars):\n")
print(all_briefs)
````

### Step 2: LLM Nominates Candidates

The LLM reviews the brief summaries and selects which tables are likely relevant to the question. This simulates what the `agent_context_prefilter` agent does with the `nominate_tables` tool.

````python
from google import genai
from google.genai import types as genai_types
from config import TOOL_MODEL_LOCATION

client = genai.Client(
    vertexai=True,
    project=GOOGLE_CLOUD_PROJECT,
    location=TOOL_MODEL_LOCATION or "us-central1",
)

nomination_prompt = f"""\
Review the brief metadata summaries below and identify which tables are
relevant to answering the user's question. Return a JSON array of fully
qualified table IDs (project.dataset.table). Be generous — include
borderline tables since a downstream reranker will sort by relevance.

## Question
{QUESTION}

## Table Summaries
{all_briefs}"""

response = client.models.generate_content(
    model=TOOL_MODEL,
    contents=nomination_prompt,
    config=genai_types.GenerateContentConfig(
        response_mime_type="application/json",
        temperature=0.1,
    ),
)

nominated_ids = json.loads(response.text)
print(f"LLM nominated {len(nominated_ids)} table(s):\n")
for tid in nominated_ids:
    print(f"  {tid}")
````

### Step 3: Rerank Nominees

Fetch the detailed cached context (full YAML capsules with data profiles) for only the nominated tables and pass to the reranker.

````python
prefilter_detailed = get_detailed_for_tables(nominated_ids)
print(f"Detailed context for {len(nominated_ids)} nominees: {len(prefilter_detailed):,} chars")
print(f"  (vs {len(dataplex_context):,} chars for all tables in Approach 3)\n")

result_context_prefilter = call_reranker(
    question=QUESTION,
    candidate_metadata=prefilter_detailed,
    discovery_method="context_prefilter",
    top_k=TOP_K,
)

print(f"Ranked {len(result_context_prefilter.ranked_tables)} tables:\n")
for t in result_context_prefilter.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    print()
````

````python
print(result_context_prefilter.model_dump_json(indent=2))
````

## Approach 5: Semantic Context

This approach combines Dataplex semantic search (like Approach 2) with cached Knowledge Context capsules (like Approach 3). It uses semantic search to narrow down to ~20 candidates, then looks up their detailed cached context instead of making per-table `lookup_entry` API calls.

Fully deterministic — the entire flow runs in a callback with no LLM reasoning.

**ADK Agent:** `agent_semantic_context` — the `before_agent_callback` runs semantic search, cache lookup, and reranking. Per-query cost: 1 `search_entries` call + 0 `lookup_entry` calls (vs Approach 2's 1 + N).

### Step 1: Semantic Search + Cache Lookup

Same semantic search as Approach 2, but instead of calling `lookup_entry` per table, we look up the detailed context from the shared cache.

````python
# Semantic search — scoped to our datasets (same as Approach 2)
dataset_filter = "|".join(f"datasets/{ds}" for ds in get_datasets())
query = f"({QUESTION}) AND system=BIGQUERY AND parent:({dataset_filter})"
print(f"Search query:\n  {query}\n")

request = dataplex_v1.SearchEntriesRequest(
    name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/global",
    query=query,
    page_size=20,
    semantic_search=True,
)

matched_ids = []
skipped = 0
for result in catalog_client.search_entries(request=request):
    entry = result.dataplex_entry
    source = entry.entry_source
    fqn = entry.fully_qualified_name or ""
    parts = fqn.rsplit(".", 2)
    if len(parts) >= 2:
        ds_name, tbl_name = parts[-2], parts[-1]
        if not is_table_in_scope(ds_name, tbl_name):
            skipped += 1
            continue
        full_id = f"{GOOGLE_CLOUD_PROJECT}.{ds_name}.{tbl_name}"
        matched_ids.append(full_id)
        print(f"  Found: {source.display_name if source else entry.name}")

print(f"\nIn-scope results: {len(matched_ids)} (filtered out {skipped})")

# Cache lookup instead of lookup_entry
semantic_detailed = get_detailed_for_tables(matched_ids)
print(f"\nDetailed context from cache: {len(semantic_detailed):,} chars")
print(f"  Approach 2 makes {len(matched_ids)} lookup_entry API calls for this")
print(f"  Approach 5 makes 0 — all from cache")
````

### Step 2: Rerank

````python
result_semantic_context = call_reranker(
    question=QUESTION,
    candidate_metadata=semantic_detailed,
    discovery_method="semantic_context",
    top_k=TOP_K,
)

print(f"Ranked {len(result_semantic_context.ranked_tables)} tables:\n")
for t in result_semantic_context.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    print()
````

## Compare Results

Now let's compare the results from all five approaches side-by-side. Each approach has two stages:
1. **Nominations** — the tables discovered and sent to the reranker
2. **Reranker output** — the tables that survived reranking with confidence scores

````python
results = {
    "bq_tools": result_bq_tools,
    "dataplex_search": result_dataplex_search,
    "dataplex_context": result_dataplex_context,
    "context_prefilter": result_context_prefilter,
    "semantic_context": result_semantic_context,
}

# Nominations: tables sent TO the reranker by each approach
nominations = {
    "bq_tools": set(bq_nominated_ids),
    "dataplex_search": set(search_nominated_ids),
    "dataplex_context": set(get_table_ids()),
    "context_prefilter": set(nominated_ids),
    "semantic_context": set(matched_ids),
}

approach_labels = ["BQ Tools", "Dataplex S", "Dataplex C", "PreFilter", "Semantic C"]
approach_keys = ["bq_tools", "dataplex_search", "dataplex_context", "context_prefilter", "semantic_context"]

# --- Nominations Table ---
all_nominated = set()
for ids in nominations.values():
    all_nominated |= ids

print(f"Question: {QUESTION}")
print(f"\n{'='*120}")
print(f"NOMINATIONS — tables sent to the reranker ({len(all_nominated)} unique)")
print(f"{'='*120}")

header = f"{'Table ID':<60} " + " ".join(f"{l:<14}" for l in approach_labels)
print(header)
print("-" * len(header))

for table_id in sorted(all_nominated):
    cols = []
    for key in approach_keys:
        cols.append("nominated" if table_id in nominations[key] else "--")
    print(f"{table_id:<60} " + " ".join(f"{c:<14}" for c in cols))

for key, label in zip(approach_keys, approach_labels):
    print(f"  {label}: {len(nominations[key])} nominated")

# --- Reranker Results Table ---
all_ranked = set()
for result in results.values():
    for t in result.ranked_tables:
        all_ranked.add(t.table_id)

print(f"\n{'='*120}")
print(f"RERANKER OUTPUT — tables that survived reranking ({len(all_ranked)} unique)")
print(f"{'='*120}")

print(header)
print("-" * len(header))

for table_id in sorted(all_ranked):
    cols = []
    for key in approach_keys:
        match = next((t for t in results[key].ranked_tables if t.table_id == table_id), None)
        if match:
            cols.append(f"#{match.rank} ({match.confidence:.2f})")
        else:
            cols.append("--")
    print(f"{table_id:<60} " + " ".join(f"{c:<14}" for c in cols))

# --- Filtering Summary ---
print(f"\n{'='*120}")
print("FILTERING SUMMARY — nominations vs reranker output")
print(f"{'='*120}")
for key, label in zip(approach_keys, approach_labels):
    ranked_ids = {t.table_id for t in results[key].ranked_tables}
    filtered_out = nominations[key] - ranked_ids
    print(f"  {label}: {len(nominations[key])} nominated → {len(ranked_ids)} ranked ({len(filtered_out)} filtered out)")
    if filtered_out:
        for tid in sorted(filtered_out):
            print(f"    dropped: {tid}")
````

### Agreement and Unique Finds

````python
# Tables ranked by all five approaches
tables_by_approach = {}
for name, result in results.items():
    tables_by_approach[name] = {t.table_id for t in result.ranked_tables}

all_five = set.intersection(*tables_by_approach.values()) if tables_by_approach else set()
print(f"Ranked by ALL five approaches ({len(all_five)}):")
for t in sorted(all_five):
    scores = []
    for key in approach_keys:
        match = next((r for r in results[key].ranked_tables if r.table_id == t), None)
        scores.append(f"{match.confidence:.2f}" if match else "--")
    print(f"  {t}: confidence = {' / '.join(scores)}")

# Tables nominated by all five but dropped by some rerankers
nominated_by_all = set.intersection(*nominations.values()) if nominations else set()
nominated_but_not_ranked = nominated_by_all - all_five
if nominated_but_not_ranked:
    print(f"\nNominated by ALL five but dropped by at least one reranker ({len(nominated_but_not_ranked)}):")
    for t in sorted(nominated_but_not_ranked):
        dropped_by = [label for key, label in zip(approach_keys, approach_labels) if t not in tables_by_approach[key]]
        print(f"  {t} — dropped by: {', '.join(dropped_by)}")

# Unique finds per approach
print()
for name in approach_keys:
    others = set()
    for other_name, other_tables in tables_by_approach.items():
        if other_name != name:
            others |= other_tables
    unique = tables_by_approach[name] - others
    if unique:
        print(f"Unique to {name}: {sorted(unique)}")
    else:
        print(f"Unique to {name}: (none)")
````

## Run as ADK Agents

The approaches above are packaged as [ADK](https://google.github.io/adk-docs/) agents — five total, including two additional approaches (Context Pre-Filter and Semantic Context) that use a shared context cache. They can run independently or together via a `ParallelAgent` orchestrator.

**Interactive web UI (recommended):**
```
uv run adk web .
```

Use the agent selector dropdown (top-left) to choose which agent to run — `agent_orchestrator` runs all five in parallel, or pick any individual approach to run it standalone.

The orchestrator uses a `SequentialAgent` → `ParallelAgent` → compare agent pipeline:

```
User question
    │
    ▼
SequentialAgent (orchestrator)
    │
    ├──▶ ParallelAgent ──────────────────────────────────────────────────┐
    │       │                                                            │
    │       ├── agent_bq_tools (LLM-driven)                              │
    │       │     LLM reasoning loop:                                    │
    │       │       list_dataset_ids → list_table_ids → get_table_info   │
    │       │       → rerank_tables tool (calls call_reranker)           │
    │       │                                                            │
    │       ├── agent_dataplex_search (callback-driven, deterministic)    │
    │       │     before_agent_callback:                                  │
    │       │       search_entries (semantic) → lookup_entry per table    │
    │       │       → call_reranker → return Content (LLM skipped)       │
    │       │                                                            │
    │       ├── agent_dataplex_context (callback-driven, deterministic)  │
    │       │     before_agent_callback:                                  │
    │       │       read shared cache (pre-fetched at startup)            │
    │       │       → call_reranker → return Content (LLM skipped)       │
    │       │                                                            │
    │       ├── agent_context_prefilter (LLM + callback hybrid)          │
    │       │     LLM reviews brief summaries → calls nominate_tables    │
    │       │     after_agent_callback:                                   │
    │       │       get detailed cache for nominees → call_reranker       │
    │       │                                                            │
    │       └── agent_semantic_context (callback-driven, deterministic)  │
    │             before_agent_callback:                                  │
    │               search_entries (semantic) → cache lookup per match    │
    │               → call_reranker → return Content (LLM skipped)       │
    │                                                                    │
    │    ◀──────────────────────────────────────────────────────────────┘
    │
    └──▶ compare_results (LLM synthesizes all five from state)
```

All five approaches use the same `call_reranker` function (Gemini structured output) to produce a `RerankerResponse` with ranked tables, confidence scores, column hints, and SQL suggestions. Approaches 2, 3, and 5 run entirely in `before_agent_callback` — no LLM agent calls needed. Approach 4 uses LLM reasoning to review briefs with deterministic reranking in `after_agent_callback`. Approach 1 uses LLM reasoning to iterate through BigQuery metadata tools.
