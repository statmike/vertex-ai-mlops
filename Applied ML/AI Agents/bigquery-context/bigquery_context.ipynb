# BigQuery Context — Three Approaches to Table Discovery

This notebook walks through three different approaches to finding the best BigQuery tables for answering a user's question. All three approaches:

1. Discover candidate tables using different metadata sources
2. Feed the metadata to a shared **Gemini-powered reranker** that produces structured output
3. Return a `RerankerResponse` with ranked tables, confidence scores, column hints, and SQL suggestions

By running all three approaches against the same question, you can compare:
- Which tables each approach finds
- How much metadata each approach provides
- How confidence scores and SQL hints differ
- The trade-offs between completeness, speed, and metadata richness

**As ADK Agents**: These three approaches are also packaged as [ADK](https://google.github.io/adk-docs/) agents that run in parallel via a `ParallelAgent`. See the [readme](readme.md) for details on running them with `uv run adk web` (interactive) or `uv run adk run` (CLI).

## Approach 1: BQ Metadata Tools

This approach uses the BigQuery API to enumerate datasets and tables, then inspects schemas to gather metadata. It's the most straightforward method — systematically walk through each dataset, list tables, and read their schemas.

**ADK Agent:** `agent_bq_tools` — uses the ADK `BigQueryToolset` (filtered to `list_dataset_ids`, `list_table_ids`, `get_dataset_info`, `get_table_info`). The agent makes **per-query API calls** — each user question triggers a fresh enumeration of datasets and tables to gather metadata, which is then passed to the shared reranker.

### Step 1: List Datasets in Scope

````python
datasets = get_datasets()
print(f"Datasets in scope ({len(datasets)}):")
for ds in datasets:
    dataset = bq_client.get_dataset(f"{GOOGLE_CLOUD_PROJECT}.{ds}")
    print(f"  {ds}: {dataset.description}")
````

### Step 2: List Tables in Each Dataset

````python
all_tables = {}
for ds in datasets:
    scoped = get_scoped_tables(ds)
    if scoped is None:
        # All tables — discover via BQ API
        tables = list(bq_client.list_tables(f"{GOOGLE_CLOUD_PROJECT}.{ds}"))
        all_tables[ds] = [t.table_id for t in tables]
        print(f"  {ds}: {all_tables[ds]} (all)")
    else:
        # Specific tables from scope
        all_tables[ds] = scoped
        print(f"  {ds}: {all_tables[ds]} (scoped)")

print(f"\nTotal tables: {sum(len(v) for v in all_tables.values())}")
````

### Step 3: Get Table Details

For each table, retrieve the full schema with column names, types, and descriptions.

````python
bq_metadata_parts = []

for ds, table_ids in all_tables.items():
    for table_id in table_ids:
        full_id = f"{GOOGLE_CLOUD_PROJECT}.{ds}.{table_id}"
        table = bq_client.get_table(full_id)

        part = f"## {full_id}\n"
        part += f"Description: {table.description or '(none)'}\n"
        part += f"Rows: {table.num_rows:,}\n"
        part += f"Columns:\n"
        for field in table.schema:
            desc = f" - {field.description}" if field.description else ""
            part += f"  - {field.name} ({field.field_type}){desc}\n"

        bq_metadata_parts.append(part)

        # Show what's being gathered
        print(f"  {full_id}")
        print(f"    Description: {(table.description or '(none)')[:80]}")
        print(f"    Columns ({len(table.schema)}): {', '.join(f.name for f in table.schema[:5])}", end="")
        if len(table.schema) > 5:
            print(f" ... +{len(table.schema) - 5} more")
        else:
            print()
        cols_with_desc = sum(1 for f in table.schema if f.description)
        print(f"    Column descriptions: {cols_with_desc}/{len(table.schema)}")
        print(f"    Metadata size: {len(part):,} chars")
        print()

bq_candidate_metadata = "\n".join(bq_metadata_parts)
print(f"Total metadata for reranker: {len(bq_candidate_metadata):,} chars across {len(bq_metadata_parts)} tables")
````

### Step 4: Rerank with Gemini Structured Output

Pass all gathered metadata to the reranker, which uses Gemini to produce a `RerankerResponse` with ranked tables, confidence scores, column hints, and SQL suggestions.

````python
result_bq_tools = call_reranker(
    question=QUESTION,
    candidate_metadata=bq_candidate_metadata,
    discovery_method="bq_tools",
    top_k=TOP_K,
)

print(f"Ranked {len(result_bq_tools.ranked_tables)} tables:\n")
for t in result_bq_tools.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    print()
````

````python
print(result_bq_tools.model_dump_json(indent=2))
````

## Approach 2: Dataplex Catalog Search

This approach uses Dataplex's semantic search to find relevant tables using natural language. Instead of enumerating every table, it sends the user's question to `search_entries` with `semantic_search=True`, which uses embeddings to find tables whose names, descriptions, and metadata relate to the query. Results are post-filtered to the configured scope, then each match is looked up for full schema details.

**ADK Agent:** `agent_catalog_search` — uses custom `search_catalog` and `lookup_entry` tools. The agent makes **per-query API calls** — each user question triggers a new semantic search and entry lookups, with results passed to the shared reranker. This is the only approach where the discovery step itself is query-dependent (the search is driven by the question's semantics).

### Step 1: Semantic Search via Dataplex

````python
catalog_client = dataplex_v1.CatalogServiceClient()

# Semantic search: use natural language query with system filter only.
# The linked_resource filter is not effective with semantic_search=True
# (Dataplex interprets the entire query semantically), so we post-filter
# results using is_table_in_scope instead.
query = f"({QUESTION}) AND system=BIGQUERY"
print(f"Search query:\n  {query}\n")

request = dataplex_v1.SearchEntriesRequest(
    name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/global",
    query=query,
    page_size=20,
    semantic_search=True,
)

search_results = []
skipped = 0
for result in catalog_client.search_entries(request=request):
    entry = result.dataplex_entry
    source = entry.entry_source
    fqn = entry.fully_qualified_name or ""

    # Post-filter to only tables in our configured scope
    parts = fqn.rsplit(".", 2)
    if len(parts) >= 2:
        ds_name, tbl_name = parts[-2], parts[-1]
        if not is_table_in_scope(ds_name, tbl_name):
            skipped += 1
            continue

    search_results.append({
        "entry_name": entry.name,
        "fully_qualified_name": fqn,
        "display_name": source.display_name if source else "",
        "description": source.description if source else "",
    })
    print(f"  Found: {source.display_name if source else entry.name}")

print(f"\nIn-scope results: {len(search_results)} (filtered out {skipped} from other datasets)")
````

### Step 2: Lookup Entry Details

For each search result, retrieve full metadata including schema, storage info, and any data profile aspects.

````python
catalog_metadata_parts = []

for sr in search_results:
    request = dataplex_v1.LookupEntryRequest(
        name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/{BQ_LOCATION.lower()}",
        entry=sr["entry_name"],
        view=dataplex_v1.EntryView.FULL,
    )
    entry = catalog_client.lookup_entry(request=request)
    entry_dict = json_format.MessageToDict(entry._pb)

    # Extract key metadata
    source = entry_dict.get("entrySource", {})
    aspects = entry_dict.get("aspects", {})

    # Parse fully_qualified_name to project.dataset.table format
    # FQN format: bigquery:project.dataset.table
    fqn = sr.get("fully_qualified_name", "")
    fqn_clean = fqn.split(":", 1)[-1] if ":" in fqn else fqn
    table_label = fqn_clean or sr["display_name"]

    part = f"## {table_label}\n"
    part += f"Description: {source.get('description', '(none)')}\n"

    # Schema
    for key, aspect in aspects.items():
        if "schema" in key:
            schema_data = aspect.get("data", {})
            fields = schema_data.get("fields", [])
            part += f"Columns ({len(fields)}):\n"
            for fld in fields:
                desc = fld.get("description", "")
                part += f"  - {fld.get('name', '?')} ({fld.get('type', '?')}) {desc}\n"

    catalog_metadata_parts.append(part)
    print(f"  {table_label}: {len(part):,} chars of metadata")

catalog_candidate_metadata = "\n".join(catalog_metadata_parts)
print(f"\nTotal metadata: {len(catalog_candidate_metadata):,} chars")
````

### Step 3: Rerank

````python
if not catalog_candidate_metadata.strip():
    print("No in-scope tables found by semantic search — skipping reranker.")
    print("This can happen if Dataplex hasn't indexed the tables yet.")
    print("Try re-running after a few minutes.")
    result_catalog_search = RerankerResponse(question=QUESTION, top_k=TOP_K, ranked_tables=[], notes="No in-scope results from semantic search.")
else:
    result_catalog_search = call_reranker(
        question=QUESTION,
        candidate_metadata=catalog_candidate_metadata,
        discovery_method="catalog_search",
        top_k=TOP_K,
    )

    print(f"Ranked {len(result_catalog_search.ranked_tables)} tables:\n")
    for t in result_catalog_search.ranked_tables:
        print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
        print(f"     {t.reasoning}")
        print()
````

#### Full Reranker Output

````python
print(result_catalog_search.model_dump_json(indent=2))
````

## Approach 3: Knowledge Context API

This approach uses the Dataplex `lookupContext` API to fetch pre-formatted, LLM-ready metadata capsules for all tables at once. The capsules include schema, column descriptions, and — when data profiling has been run — column-level statistics like null ratios, distinct values, and sample values.

> **Note:** The `lookupContext` API is not yet in the `google-cloud-dataplex` Python SDK (v2.16.0). This notebook calls it via REST using `google.auth` credentials.

**ADK Agent:** `agent_knowledge_context` — uses the `initialize_context` tool which calls `lookupContext` and caches the result in agent state. Unlike the other two approaches, metadata is fetched **once at startup** (zero per-query cost). When a user question arrives, the agent passes the pre-cached context directly to the shared reranker — no additional API calls are made. This makes it the fastest approach at query time, and the richest in metadata when Dataplex profiling has been run.

### Step 1: Fetch Context Capsules

Fetch both dataset-level context (descriptions) and table-level context (schema + data profiles) in batched API calls.

````python
# Build table entry names based on scope, tracking project.dataset.table labels
datasets = get_datasets()
table_entries = []  # list of (full_bq_id, entry_name)
for ds in datasets:
    scoped = get_scoped_tables(ds)
    if scoped is None:
        # All tables — discover via BQ API
        for t in bq_client.list_tables(f"{GOOGLE_CLOUD_PROJECT}.{ds}"):
            full_id = f"{GOOGLE_CLOUD_PROJECT}.{ds}.{t.table_id}"
            table_entries.append((full_id, get_dataplex_entry_name(ds, t.table_id)))
    else:
        # Specific tables — use directly
        for table_name in scoped:
            full_id = f"{GOOGLE_CLOUD_PROJECT}.{ds}.{table_name}"
            table_entries.append((full_id, get_dataplex_entry_name(ds, table_name)))
print(f"Resolved {len(table_entries)} tables across {len(datasets)} datasets")

# Dataset-level context
dataset_entry_names = [get_dataplex_dataset_entry_name(ds) for ds in datasets]
print(f"\nFetching context for {len(dataset_entry_names)} datasets...")
dataset_context = lookup_context_batched(dataset_entry_names)
print(f"  Dataset context: {len(dataset_context):,} chars")

# Table-level context with project.dataset.table headers
# Call per-table so we can prepend the standard BQ table ID
from agent_knowledge_context.tools.util_lookup_context import lookup_context
print(f"Fetching context for {len(table_entries)} tables...")
table_context_parts = []
for full_id, entry_name in table_entries:
    ctx = lookup_context([entry_name])
    if ctx:
        table_context_parts.append(f"## {full_id}\n{ctx}")
table_context = "\n\n".join(table_context_parts)
print(f"  Table context: {len(table_context):,} chars")

# Combined
knowledge_context = f"# Dataset Context\n{dataset_context}\n\n# Table Context\n{table_context}"
print(f"\nTotal knowledge context: {len(knowledge_context):,} chars")
````

#### Preview: Knowledge Context Capsule

This is what the `lookupContext` API returns — LLM-ready metadata in YAML format. Notice how data profiling adds column-level statistics (null ratios, distinct values, sample values).

````python
# Show first 3000 chars of the knowledge context
preview = knowledge_context[:3000]
print(preview)
if len(knowledge_context) > 3000:
    print(f"\n... ({len(knowledge_context) - 3000:,} more chars)")
````

### Step 2: Rerank

````python
result_knowledge_context = call_reranker(
    question=QUESTION,
    candidate_metadata=knowledge_context,
    discovery_method="knowledge_context",
    top_k=TOP_K,
)

print(f"Ranked {len(result_knowledge_context.ranked_tables)} tables:\n")
for t in result_knowledge_context.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    print()
````

#### Full Reranker Output

````python
print(result_knowledge_context.model_dump_json(indent=2))
````

## Compare Results

Now let's compare the results from all three approaches side-by-side.

````python
results = {
    "bq_tools": result_bq_tools,
    "catalog_search": result_catalog_search,
    "knowledge_context": result_knowledge_context,
}

# Collect all unique tables across approaches
all_found_tables = set()
for name, result in results.items():
    for t in result.ranked_tables:
        all_found_tables.add(t.table_id)

print(f"Question: {QUESTION}")
print(f"Unique tables found across all approaches: {len(all_found_tables)}\n")

# Side-by-side comparison
print(f"{'Table ID':<60} {'BQ Tools':<15} {'Catalog':<15} {'Knowledge':}")
print("=" * 105)

for table_id in sorted(all_found_tables):
    cols = []
    for name in ["bq_tools", "catalog_search", "knowledge_context"]:
        match = next((t for t in results[name].ranked_tables if t.table_id == table_id), None)
        if match:
            cols.append(f"#{match.rank} ({match.confidence:.2f})")
        else:
            cols.append("--")
    print(f"{table_id:<60} {cols[0]:<15} {cols[1]:<15} {cols[2]}")
````

### Agreement and Unique Finds

````python
# Tables found by all three approaches
tables_by_approach = {}
for name, result in results.items():
    tables_by_approach[name] = {t.table_id for t in result.ranked_tables}

all_three = tables_by_approach["bq_tools"] & tables_by_approach["catalog_search"] & tables_by_approach["knowledge_context"]
print(f"Found by ALL three approaches ({len(all_three)}):")
for t in sorted(all_three):
    scores = []
    for name in ["bq_tools", "catalog_search", "knowledge_context"]:
        match = next((r for r in results[name].ranked_tables if r.table_id == t), None)
        scores.append(f"{match.confidence:.2f}" if match else "--")
    print(f"  {t}: confidence = {' / '.join(scores)}")

# Unique finds per approach
print()
for name in ["bq_tools", "catalog_search", "knowledge_context"]:
    others = set()
    for other_name, other_tables in tables_by_approach.items():
        if other_name != name:
            others |= other_tables
    unique = tables_by_approach[name] - others
    if unique:
        print(f"Unique to {name}: {sorted(unique)}")
    else:
        print(f"Unique to {name}: (none)")
````

### SQL Hints Comparison

Compare the quality of SQL guidance from each approach for the top-ranked table.

````python
for name, result in results.items():
    if result.ranked_tables:
        top = result.ranked_tables[0]
        print(f"--- {name} (top table: {top.table_id}) ---")
        print(f"SQL hints: {top.sql_hints}")
        print(f"Key columns: {[c.name for c in top.key_columns]}")
        if top.join_suggestions:
            print(f"Join suggestions: {[(j.target_table, j.join_keys) for j in top.join_suggestions]}")
        print()
````

### Metadata Richness

Compare the amount of metadata each approach provided to the reranker.

````python
print(f"Metadata provided to reranker:")
print(f"  BQ Tools:          {len(bq_candidate_metadata):>8,} chars (schema + row counts)")
print(f"  Catalog Search:    {len(catalog_candidate_metadata):>8,} chars (schema + aspects)")
print(f"  Knowledge Context: {len(knowledge_context):>8,} chars (schema + profiles + descriptions)")
print()
print("The Knowledge Context approach typically provides the richest metadata")
print("because it includes data profile statistics (null ratios, distinct values,")
print("sample values) published by Dataplex data profiling.")
````

```
uv run adk run agent_knowledge_context     # Approach 3 only
```

The orchestrator uses a `SequentialAgent` → `ParallelAgent` → compare agent pipeline:

```
User question
    │
    ▼
SequentialAgent (orchestrator)
    │
    ├──▶ ParallelAgent ──────────────────────────────────────────────────┐
    │       │                                                            │
    │       ├── agent_bq_tools (LLM-driven)                              │
    │       │     LLM reasoning loop:                                    │
    │       │       list_dataset_ids → list_table_ids → get_table_info   │
    │       │       → rerank_tables tool (calls call_reranker)           │
    │       │                                                            │
    │       ├── agent_catalog_search (callback-driven, deterministic)    │
    │       │     before_agent_callback:                                  │
    │       │       search_entries (semantic) → lookup_entry per table    │
    │       │       → call_reranker → return Content (LLM skipped)       │
    │       │                                                            │
    │       └── agent_knowledge_context (callback-driven, deterministic) │
    │             before_agent_callback:                                  │
    │               read module-level cache (pre-fetched at startup)      │
    │               → call_reranker → return Content (LLM skipped)       │
    │                                                                    │
    │    ◀──────────────────────────────────────────────────────────────┘
    │
    └──▶ compare_results (LLM synthesizes all three from state)
```

All three approaches use the same `call_reranker` function (Gemini structured output) to produce a `RerankerResponse` with ranked tables, confidence scores, column hints, and SQL suggestions. Approaches 2 and 3 run entirely in `before_agent_callback` — no LLM agent calls needed — while approach 1 uses LLM reasoning to iterate through BigQuery metadata tools.
