# BigQuery Context — Five Approaches to Table Discovery

This notebook is a hands-on guide to understanding five different approaches for finding the best BigQuery tables to answer a user's question. Rather than running the ADK agents directly, we **reproduce each approach step-by-step** so you can see exactly:

1. **What metadata each API returns** — for the same table, side by side
2. **What gets nominated** — which tables each approach sends to the reranker, and in what format
3. **What the reranker receives** — the actual metadata string, so you can see the richness difference
4. **What the reranker outputs** — structured results with confidence scores, column hints, and SQL suggestions

All five approaches use the same shared reranker ([Gemini structured output](https://ai.google.dev/gemini-api/docs/structured-output) → `RerankerResponse`), so the only variable is *how tables are discovered* and *what metadata describes them*.

**To run the agents interactively**, see the [readme](readme.md) — use `uv run adk web .` to launch the ADK web UI.

````python
QUESTION = "Which US counties have the most weather stations per capita?"
print(f"Question: {QUESTION}")

# We'll use weather_stations as the example table throughout The APIs section
EXAMPLE_TABLE = "weather_stations"
EXAMPLE_DATASET = "bigquery_context_weather"
EXAMPLE_FULL_ID = f"{GOOGLE_CLOUD_PROJECT}.{EXAMPLE_DATASET}.{EXAMPLE_TABLE}"
print(f"Example table: {EXAMPLE_FULL_ID}")
````

## The APIs — Same Table, Different Metadata

Before walking through the five approaches, let's see what metadata each API returns for the **same table** (`weather_stations`). This makes the richness differences concrete.

| API | SDK / Protocol | Used By | Key Difference |
|---|---|---|---|
| BQ `get_table_info` | BigQuery Python SDK | Approach 1 | Schema + row count, no column descriptions or profiling |
| Dataplex `lookup_entry` | `google-cloud-dataplex` SDK | Approach 2 | Schema + catalog aspects, no data profiling stats |
| Dataplex `lookupContext` | REST API (not yet in SDK) | Approaches 3, 4, 5 (via cache) | Schema + `dataProfile` per column (nullRatio, distinctValues, sampleValues) |

### BigQuery API: `get_table_info`

This is what Approach 1 works with — the standard BigQuery SDK. It returns column names, types, and modes, plus table description and row count. Column descriptions come through when they exist in the BQ schema, but there are no data profiling statistics.

````python
table = bq_client.get_table(EXAMPLE_FULL_ID)

bq_metadata = f"## {EXAMPLE_FULL_ID}\n"
bq_metadata += f"Description: {table.description or '(none)'}\n"
bq_metadata += f"Rows: {table.num_rows:,}\n"
bq_metadata += f"Columns:\n"
for field in table.schema:
    desc = f" - {field.description}" if field.description else ""
    bq_metadata += f"  - {field.name} ({field.field_type}){desc}\n"

print(bq_metadata)
print(f"\n--- {len(bq_metadata):,} chars ---")
````

### Dataplex SDK: `lookup_entry`

This is what Approach 2 works with — the Dataplex Python SDK's [`lookup_entry`](https://cloud.google.com/python/docs/reference/dataplex/latest/google.cloud.dataplex_v1.services.catalog_service.CatalogServiceClient#google_cloud_dataplex_v1_services_catalog_service_CatalogServiceClient_lookup_entry). It returns schema with column names/types/descriptions, entry source metadata, and catalog aspects. No data profiling statistics — the metadata is richer than BQ's but still schema-level only.

````python
# Use the same entry name builder as the agents
entry_name = get_dataplex_entry_name(EXAMPLE_DATASET, EXAMPLE_TABLE)
print(f"Entry name: {entry_name}\n")

lookup_req = dataplex_v1.LookupEntryRequest(
    name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/{BQ_LOCATION.lower()}",
    entry=entry_name,
    view=dataplex_v1.EntryView.FULL,
)
entry = catalog_client.lookup_entry(request=lookup_req)
entry_dict = json_format.MessageToDict(entry._pb)

# Build the summary the same way approach 2's callback does
lookup_entry_metadata = {
    "display_name": entry_dict.get("entrySource", {}).get("displayName", ""),
    "description": entry_dict.get("entrySource", {}).get("description", ""),
}
for key, aspect in entry_dict.get("aspects", {}).items():
    if "schema" in key:
        lookup_entry_metadata["schema"] = aspect.get("data", {})

lookup_entry_json = json.dumps(lookup_entry_metadata, indent=2)
print(f"## {EXAMPLE_FULL_ID}")
print(lookup_entry_json)
print(f"\n--- {len(lookup_entry_json):,} chars ---")
````

### Dataplex REST API: `lookupContext`

This is what Approaches 3, 4, and 5 work with (via the shared cache). The [`lookupContext`](https://cloud.google.com/dataplex/docs/reference/rest/v1/projects.locations/lookupContext) REST API returns LLM-ready metadata capsules in JSON format. When Dataplex data profiling has been run, each column includes a `dataProfile` section with `nullRatio`, `distinctValues`, and `sampleValues`.

> **Note:** This API is not yet in the `google-cloud-dataplex` Python SDK. We call it via REST with `google.auth` credentials.

````python
dataplex_entry_name = get_dataplex_entry_name(EXAMPLE_DATASET, EXAMPLE_TABLE)
lookup_context_json = lookup_context([dataplex_entry_name])

# Parse and pretty-print
lookup_context_parsed = json.loads(lookup_context_json)
lookup_context_pretty = json.dumps(lookup_context_parsed, indent=2)
print(lookup_context_pretty)
print(f"\n--- {len(lookup_context_pretty):,} chars ---")
````

### Shared Context Cache: Brief vs Detailed

The shared cache (`context_cache/`) is populated once at startup from the `lookupContext` API. It stores two views per table — both derived from the same JSON response:

- **Brief**: `dataProfile` stripped — just table name, description, and schema columns (name, type, description). Used by Approach 4 for LLM pre-filtering.
- **Detailed**: Full JSON including `dataProfile` per column. Used by the reranker in Approaches 3, 4, and 5.

Let's see both views for our example table:

````python
# Brief view — what the LLM sees in Approach 4 for pre-filtering
all_briefs_parsed = json.loads(get_all_briefs())
example_brief = next((b for b in all_briefs_parsed if b.get("table_id") == EXAMPLE_FULL_ID), None)
if example_brief:
    brief_str = json.dumps(example_brief, indent=2)
    print("=== BRIEF (no dataProfile) ===")
    print(brief_str)
    print(f"\n--- {len(brief_str):,} chars ---")
````

````python
# Detailed view — what the reranker sees in Approaches 3, 4, 5
detailed_for_example = get_detailed_for_tables([EXAMPLE_FULL_ID])
detailed_parsed = json.loads(detailed_for_example)
if detailed_parsed:
    detailed_str = json.dumps(detailed_parsed[0], indent=2)
    print("=== DETAILED (with dataProfile) ===")
    print(detailed_str)
    print(f"\n--- {len(detailed_str):,} chars ---")
````

````python
# Side-by-side: metadata size for the same table across APIs
print(f"Metadata size for {EXAMPLE_TABLE} across APIs:")
print(f"  BQ get_table_info:      {len(bq_metadata):>6,} chars  (schema + row count)")
print(f"  Dataplex lookup_entry:  {len(lookup_entry_json):>6,} chars  (schema + aspects)")
print(f"  Cache brief:            {len(brief_str):>6,} chars  (lookupContext minus dataProfile)")
print(f"  Cache detailed:         {len(detailed_str):>6,} chars  (full lookupContext with dataProfile)")
print()
print(f"The detailed view is {len(detailed_str)/len(bq_metadata):.1f}x richer than BQ metadata.")
print(f"The brief view is {len(brief_str)/len(bq_metadata):.1f}x the size of BQ metadata.")
print()
print("Key additions in lookupContext vs BQ/lookup_entry:")
print("  - dataProfile.nullRatio — what fraction of values are null")
print("  - dataProfile.distinctValues — cardinality as a percentage")
print("  - dataProfile.sampleValues — actual example values from the data")
````

---
## Approach 1: BQ Metadata Tools (`agent_bq_tools`)

**Discovery:** LLM-driven tool calls via ADK [BigQueryToolset](https://google.github.io/adk-docs/integrations/bigquery/) — `list_dataset_ids` → `list_table_ids` → `get_table_info` per table.
**Metadata API:** BigQuery SDK — schema (columns, types, modes), description, row count.
**Reranker called as:** ADK tool (`rerank_tables`) — the LLM decides when to call it.
**Reranker receives:** `get_table_info` text for all inspected tables (no `dataProfile`).
**Cache usage:** None.
**Per-query API calls:** 4–10+ BQ API calls.

### Step 1: Discover and Gather Metadata

The agent enumerates datasets → tables → schemas using BQ API calls. Here we reproduce that workflow:

````python
bq_metadata_parts = []
bq_nominated_ids = []

for ds in get_datasets():
    scoped = get_scoped_tables(ds)
    if scoped is None:
        table_ids = [t.table_id for t in bq_client.list_tables(f"{GOOGLE_CLOUD_PROJECT}.{ds}")]
    else:
        table_ids = scoped

    for table_id in table_ids:
        full_id = f"{GOOGLE_CLOUD_PROJECT}.{ds}.{table_id}"
        tbl = bq_client.get_table(full_id)
        bq_nominated_ids.append(full_id)

        part = f"## {full_id}\n"
        part += f"Description: {tbl.description or '(none)'}\n"
        part += f"Rows: {tbl.num_rows:,}\n"
        part += f"Columns:\n"
        for field in tbl.schema:
            desc = f" - {field.description}" if field.description else ""
            part += f"  - {field.name} ({field.field_type}){desc}\n"
        bq_metadata_parts.append(part)

        cols_with_desc = sum(1 for f in tbl.schema if f.description)
        print(f"  {full_id}: {len(tbl.schema)} cols ({cols_with_desc} with descriptions), {len(part):,} chars")

bq_candidate_metadata = "\n".join(bq_metadata_parts)
print(f"\nNominated: {len(bq_nominated_ids)} tables")
print(f"Total metadata for reranker: {len(bq_candidate_metadata):,} chars")
````

### Step 2: Preview What the Reranker Receives

The metadata string below is what Approach 1 passes to the reranker. Notice: plain text, no JSON structure, no `dataProfile`. The reranker has to work with column names, types, and descriptions only.

````python
# Show the reranker input for our example table
example_part = next(p for p in bq_metadata_parts if EXAMPLE_TABLE in p)
print("Reranker input for weather_stations (Approach 1):")
print(example_part)
````

### Step 3: Rerank

````python
result_bq_tools = call_reranker(
    question=QUESTION,
    candidate_metadata=bq_candidate_metadata,
    discovery_method="bq_tools",
    top_k=TOP_K,
)

print(f"Ranked {len(result_bq_tools.ranked_tables)} table(s):\n")
for t in result_bq_tools.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    if t.sql_hints:
        print(f"     SQL: {t.sql_hints}")
    print()
````

## Approach 2: Dataplex Search (`agent_dataplex_search`)

**Discovery:** Dataplex [`search_entries`](https://cloud.google.com/python/docs/reference/dataplex/latest/google.cloud.dataplex_v1.services.catalog_service.CatalogServiceClient#google_cloud_dataplex_v1_services_catalog_service_CatalogServiceClient_search_entries) with `semantic_search=True`, scoped to configured datasets.
**Metadata API:** Dataplex [`lookup_entry`](https://cloud.google.com/python/docs/reference/dataplex/latest/google.cloud.dataplex_v1.services.catalog_service.CatalogServiceClient#google_cloud_dataplex_v1_services_catalog_service_CatalogServiceClient_lookup_entry) per matched table — schema + catalog aspects, no `dataProfile`.
**Reranker called as:** Direct function call in `before_agent_callback` (LLM skipped).
**Reranker receives:** `lookup_entry` JSON (schema + aspects, no `dataProfile`).
**Cache usage:** None.
**Per-query API calls:** 1 search + N lookup calls.

This is the **baseline** that Approach 5 improves upon — same semantic search, but weaker metadata.

### Step 1: Semantic Search + Entry Lookup

````python
# Step 1a: Semantic search — scoped to our datasets
dataset_filter = "|".join(f"datasets/{ds}" for ds in get_datasets())
search_query = f"({QUESTION}) AND system=BIGQUERY AND parent:({dataset_filter})"
print(f"Search query:\n  {search_query}\n")

request = dataplex_v1.SearchEntriesRequest(
    name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/global",
    query=search_query,
    page_size=20,
    semantic_search=True,
)

search_results = []
for result in catalog_client.search_entries(request=request):
    entry = result.dataplex_entry
    source = entry.entry_source
    fqn = entry.fully_qualified_name or ""
    parts = fqn.rsplit(".", 2)
    if len(parts) >= 2:
        if not is_table_in_scope(parts[-2], parts[-1]):
            continue
    search_results.append({
        "entry_name": entry.name,
        "fqn": fqn,
        "display_name": source.display_name if source else "",
        "description": source.description if source else "",
    })
    print(f"  Found: {source.display_name if source else entry.name}")

print(f"\nIn-scope results: {len(search_results)}")
````

````python
# Step 1b: lookup_entry for each match — build metadata for the reranker
search_metadata_parts = []
search_nominated_ids = []

for sr in search_results:
    fqn = sr["fqn"]
    fqn_parts = fqn.rsplit(".", 2)
    table_header = f"{GOOGLE_CLOUD_PROJECT}.{fqn_parts[-2]}.{fqn_parts[-1]}" if len(fqn_parts) >= 3 else fqn

    lookup_req = dataplex_v1.LookupEntryRequest(
        name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/{BQ_LOCATION.lower()}",
        entry=sr["entry_name"],
        view=dataplex_v1.EntryView.FULL,
    )
    entry = catalog_client.lookup_entry(request=lookup_req)
    entry_dict = json_format.MessageToDict(entry._pb)

    summary = {
        "display_name": sr["display_name"],
        "description": sr["description"],
    }
    for key, aspect in entry_dict.get("aspects", {}).items():
        if "schema" in key:
            summary["schema"] = aspect.get("data", {})

    search_nominated_ids.append(table_header)
    search_metadata_parts.append(f"## {table_header}\n{json.dumps(summary, indent=2)}")
    print(f"  {table_header}: {len(search_metadata_parts[-1]):,} chars")

search_candidate_metadata = "\n\n".join(search_metadata_parts)
print(f"\nNominated: {len(search_nominated_ids)} tables")
print(f"Total metadata for reranker: {len(search_candidate_metadata):,} chars")
````

### Step 2: Preview What the Reranker Receives

Approach 2 passes `lookup_entry` JSON to the reranker. Compare this to Approach 1's plain text — it's structured JSON with nested schema, but still no `dataProfile`.

````python
# Show reranker input for our example table
example_search_part = next(p for p in search_metadata_parts if EXAMPLE_TABLE in p)
print("Reranker input for weather_stations (Approach 2):")
print(example_search_part)
````

### Step 3: Rerank

````python
result_dataplex_search = call_reranker(
    question=QUESTION,
    candidate_metadata=search_candidate_metadata,
    discovery_method="dataplex_search",
    top_k=TOP_K,
)

print(f"Ranked {len(result_dataplex_search.ranked_tables)} table(s):\n")
for t in result_dataplex_search.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    if t.sql_hints:
        print(f"     SQL: {t.sql_hints}")
    print()
````

## Approach 3: Dataplex Context (`agent_dataplex_context`)

**Discovery:** None — all tables come from the shared cache (pre-populated at startup).
**Metadata API:** Dataplex [`lookupContext`](https://cloud.google.com/dataplex/docs/reference/rest/v1/projects.locations/lookupContext) REST API (via cache) — full JSON with `dataProfile`.
**Reranker called as:** Direct function call in `before_agent_callback` (LLM skipped).
**Reranker receives:** `lookupContext` JSON for ALL cached tables — full `dataProfile` included.
**Cache usage:** `get_all_detailed()` — every table, full metadata.
**Per-query API calls:** 0 (everything from cache).

Sends the **richest metadata** but evaluates **all tables** — relies entirely on the reranker to filter.

### Step 1: Get All Detailed Context from Cache

````python
context_all_detailed = get_all_detailed()
context_nominated_ids = get_table_ids()

print(f"Nominated: {len(context_nominated_ids)} tables (all cached)")
print(f"Total metadata for reranker: {len(context_all_detailed):,} chars")
print(f"\nThis is {len(context_all_detailed)/len(bq_candidate_metadata):.1f}x the size of Approach 1's metadata")
print(f"and {len(context_all_detailed)/len(search_candidate_metadata):.1f}x the size of Approach 2's metadata")
````

### Step 2: Preview What the Reranker Receives

The reranker gets the full `lookupContext` JSON for every table — including `dataProfile` with `sampleValues`. Here's our example table in this format:

````python
# Show the detailed cached entry for our example table
example_detailed = get_detailed_for_tables([EXAMPLE_FULL_ID])
example_detailed_parsed = json.loads(example_detailed)
if example_detailed_parsed:
    print(f"Reranker input for weather_stations (Approach 3):")
    print(json.dumps(example_detailed_parsed[0], indent=2))
````

### Step 3: Rerank

````python
result_dataplex_context = call_reranker(
    question=QUESTION,
    candidate_metadata=context_all_detailed,
    discovery_method="dataplex_context",
    top_k=TOP_K,
)

print(f"Ranked {len(result_dataplex_context.ranked_tables)} table(s):\n")
for t in result_dataplex_context.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    if t.sql_hints:
        print(f"     SQL: {t.sql_hints}")
    print()
````

## Approach 4: Context Pre-Filter (`agent_context_prefilter`)

**Discovery:** LLM reviews **brief** summaries (from cache) of all tables and nominates candidates.
**Metadata API:** Dataplex [`lookupContext`](https://cloud.google.com/dataplex/docs/reference/rest/v1/projects.locations/lookupContext) REST API (via cache) — briefs for LLM, detailed for reranker.
**Reranker called as:** Direct function call in `after_agent_callback` (after LLM nominates).
**Reranker receives:** `lookupContext` JSON for **nominated tables only** — full `dataProfile`.
**Cache usage:** `get_all_briefs()` for LLM → `get_detailed_for_tables()` for reranker.
**Per-query API calls:** 0 (everything from cache).

Two-stage approach: compact briefs for LLM filtering, full metadata for reranking. The LLM sees ~1,300 chars/table (briefs), but the reranker sees ~5,500 chars/table (detailed) — only for the tables the LLM selected.

### Step 1: LLM Reviews Brief Summaries

The agent's system prompt embeds `get_all_briefs()` — every table's brief summary without `dataProfile`. The LLM reads these and decides which tables to nominate.

````python
all_briefs = get_all_briefs()
print(f"Brief summaries: {len(get_table_ids())} tables, {len(all_briefs):,} chars total")
print(f"Average: ~{len(all_briefs) // len(get_table_ids()):,} chars/table\n")

# Show the brief for our example table
example_brief_entry = next(
    (b for b in json.loads(all_briefs) if b.get("table_id") == EXAMPLE_FULL_ID), None
)
if example_brief_entry:
    print(f"Brief for {EXAMPLE_TABLE} (what the LLM sees):")
    print(json.dumps(example_brief_entry, indent=2))
````

### Step 2: LLM Nominates Candidates

We simulate the LLM's nomination step — reviewing briefs and selecting relevant tables:

````python
from google import genai
from google.genai import types as genai_types
from config import TOOL_MODEL_LOCATION

genai_client = genai.Client(
    vertexai=True,
    project=GOOGLE_CLOUD_PROJECT,
    location=TOOL_MODEL_LOCATION or "us-central1",
)

nomination_prompt = f"""\
Review the brief metadata summaries below and identify which tables are
relevant to answering the user's question. Return a JSON array of fully
qualified table IDs (project.dataset.table). Be generous — include
borderline tables since a downstream reranker will sort by relevance.

## Question
{QUESTION}

## Table Summaries
{all_briefs}"""

response = genai_client.models.generate_content(
    model=TOOL_MODEL,
    contents=nomination_prompt,
    config=genai_types.GenerateContentConfig(
        response_mime_type="application/json",
        temperature=0.0,
    ),
)

prefilter_nominated_ids = json.loads(response.text)
print(f"LLM nominated {len(prefilter_nominated_ids)} table(s):\n")
for tid in prefilter_nominated_ids:
    print(f"  {tid}")
````

### Step 3: Fetch Detailed Context for Nominees + Rerank

The `after_agent_callback` fetches full detailed context from the cache for **only** the nominated tables, then passes it to the reranker. This is the key efficiency — the reranker sees rich `dataProfile` metadata but only for the subset the LLM selected.

````python
prefilter_detailed = get_detailed_for_tables(prefilter_nominated_ids)
print(f"Detailed context for {len(prefilter_nominated_ids)} nominees: {len(prefilter_detailed):,} chars")
print(f"  vs {len(context_all_detailed):,} chars for ALL tables in Approach 3")
print(f"  → {100 * len(prefilter_detailed) / len(context_all_detailed):.0f}% of Approach 3's input\n")

result_context_prefilter = call_reranker(
    question=QUESTION,
    candidate_metadata=prefilter_detailed,
    discovery_method="context_prefilter",
    top_k=TOP_K,
)

print(f"Ranked {len(result_context_prefilter.ranked_tables)} table(s):\n")
for t in result_context_prefilter.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    if t.sql_hints:
        print(f"     SQL: {t.sql_hints}")
    print()
````

## Approach 5: Semantic Context (`agent_semantic_context`)

**Discovery:** Dataplex [`search_entries`](https://cloud.google.com/python/docs/reference/dataplex/latest/google.cloud.dataplex_v1.services.catalog_service.CatalogServiceClient#google_cloud_dataplex_v1_services_catalog_service_CatalogServiceClient_search_entries) with `semantic_search=True` (same as Approach 2).
**Metadata API:** Dataplex [`lookupContext`](https://cloud.google.com/dataplex/docs/reference/rest/v1/projects.locations/lookupContext) REST API (via cache) — full JSON with `dataProfile`.
**Reranker called as:** Direct function call in `before_agent_callback` (LLM skipped).
**Reranker receives:** `lookupContext` JSON for **matched tables only** — full `dataProfile`.
**Cache usage:** `get_detailed_for_tables()` — only search-matched tables.
**Per-query API calls:** 1 search + 0 lookups (cache replaces N `lookup_entry` calls).

The **improved version of Approach 2**: same semantic search discovery, but richer metadata from the cache instead of per-table `lookup_entry` calls.

### Step 1: Semantic Search + Cache Lookup

Same search as Approach 2, but instead of calling `lookup_entry` per table, we look up detailed context from the shared cache — zero additional API calls.

````python
# Same semantic search as Approach 2
request = dataplex_v1.SearchEntriesRequest(
    name=f"projects/{GOOGLE_CLOUD_PROJECT}/locations/global",
    query=search_query,  # reuse from Approach 2
    page_size=20,
    semantic_search=True,
)

semantic_matched_ids = []
for result in catalog_client.search_entries(request=request):
    entry = result.dataplex_entry
    fqn = entry.fully_qualified_name or ""
    parts = fqn.rsplit(".", 2)
    if len(parts) >= 2:
        if not is_table_in_scope(parts[-2], parts[-1]):
            continue
        full_id = f"{GOOGLE_CLOUD_PROJECT}.{parts[-2]}.{parts[-1]}"
        semantic_matched_ids.append(full_id)

# Cache lookup replaces N lookup_entry API calls
semantic_detailed = get_detailed_for_tables(semantic_matched_ids)

print(f"Search returned: {len(semantic_matched_ids)} in-scope tables")
print(f"Detailed context from cache: {len(semantic_detailed):,} chars")
print(f"\nApproach 2 would make {len(semantic_matched_ids)} lookup_entry API calls for this")
print(f"Approach 5 makes 0 — all from cache")
print(f"\nApproach 2 reranker input: {len(search_candidate_metadata):,} chars (lookup_entry, no dataProfile)")
print(f"Approach 5 reranker input: {len(semantic_detailed):,} chars (lookupContext, with dataProfile)")
````

### Step 2: Rerank

````python
result_semantic_context = call_reranker(
    question=QUESTION,
    candidate_metadata=semantic_detailed,
    discovery_method="semantic_context",
    top_k=TOP_K,
)

print(f"Ranked {len(result_semantic_context.ranked_tables)} table(s):\n")
for t in result_semantic_context.ranked_tables:
    print(f"  #{t.rank} [{t.confidence:.2f}] {t.table_id}")
    print(f"     {t.reasoning}")
    if t.sql_hints:
        print(f"     SQL: {t.sql_hints}")
    print()
````

## Cross-Approach Comparison

Now let's compare all five approaches side-by-side: what each nominated, what survived reranking, and how the metadata richness affected results.

````python
results = {
    "bq_tools": result_bq_tools,
    "dataplex_search": result_dataplex_search,
    "dataplex_context": result_dataplex_context,
    "context_prefilter": result_context_prefilter,
    "semantic_context": result_semantic_context,
}

nominations = {
    "bq_tools": set(bq_nominated_ids),
    "dataplex_search": set(search_nominated_ids),
    "dataplex_context": set(context_nominated_ids),
    "context_prefilter": set(prefilter_nominated_ids),
    "semantic_context": set(semantic_matched_ids),
}

metadata_sizes = {
    "bq_tools": len(bq_candidate_metadata),
    "dataplex_search": len(search_candidate_metadata),
    "dataplex_context": len(context_all_detailed),
    "context_prefilter": len(prefilter_detailed),
    "semantic_context": len(semantic_detailed),
}

approach_labels = ["1: BQ Tools", "2: Search", "3: Context", "4: PreFilter", "5: Semantic"]
approach_keys = ["bq_tools", "dataplex_search", "dataplex_context", "context_prefilter", "semantic_context"]

# Collect all tables from nominations and rankings
all_tables = set()
for ids in nominations.values():
    all_tables |= ids
for result in results.values():
    for t in result.ranked_tables:
        all_tables.add(t.table_id)

# Sort by consensus (most approaches ranking them first)
def sort_key(tid):
    ranked_count = sum(1 for key in approach_keys if any(t.table_id == tid for t in results[key].ranked_tables))
    best_conf = max(
        (next((t.confidence for t in results[key].ranked_tables if t.table_id == tid), 0.0) for key in approach_keys),
        default=0.0,
    )
    return (-ranked_count, -best_conf, tid)

sorted_tables = sorted(all_tables, key=sort_key)

# Short names for readability
def short_name(full_id):
    parts = full_id.split(".")
    return f"{parts[1]}.{parts[2]}" if len(parts) == 3 else full_id

# --- Cross-approach table ---
print(f"Question: {QUESTION}\n")
header = f"{'Table':<50} " + " ".join(f"{l:<14}" for l in approach_labels)
print(header)
print("-" * len(header))

for tid in sorted_tables:
    cols = []
    for key in approach_keys:
        ranked = next((t for t in results[key].ranked_tables if t.table_id == tid), None)
        nominated = tid in nominations[key]
        if ranked:
            cols.append(f"#{ranked.rank} ({ranked.confidence:.2f})")
        elif nominated:
            cols.append("nominated")
        else:
            cols.append("—")
    print(f"{short_name(tid):<50} " + " ".join(f"{c:<14}" for c in cols))

# --- Metadata richness ---
print(f"\n{'Reranker input size':<50} " + " ".join(f"{metadata_sizes[k]:>10,} ch  " for k in approach_keys))
print(f"{'Tables nominated':<50} " + " ".join(f"{len(nominations[k]):>10}    " for k in approach_keys))
print(f"{'Tables ranked':<50} " + " ".join(f"{len(results[k].ranked_tables):>10}    " for k in approach_keys))
````
