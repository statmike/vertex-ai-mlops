# Use Agent Deployed To Vertex AI Agent Engine

In the [deploy-vertex-ai-agent-engine](./deploy-vertex-ai-agent-engine.ipynb) workflow we deployed our agent to Vertex AI and stored the resource address in the local `.env` file. ### Use CURL:

```python
URL = f"https://{os.getenv('GOOGLE_CLOUD_LOCATION')}-aiplatform.googleapis.com/v1/{remote_app.resource_name}"
URL
```

```python
payload = dict(
    class_method = 'async_get_session',
    input = dict(
        user_id = 'u_123',
        session_id = remote_session['id']
    )
)
payload
```

```python
import google.auth
```

```python
credentials, _ = google.auth.default()
auth_req = google.auth.transport.requests.Request()
credentials.refresh(auth_req)
headers = {'Authorization': f'Bearer {credentials.token}'}
response = requests.post(URL+':query', headers = headers, json = payload)    
response.raise_for_status()
```

```python
response.json()
```

```python
payload = dict(
    class_method = 'async_stream_query',
    input = dict(
        user_id = 'u_123',
        session_id = remote_session['id'],
        message = 'Do you do data about baseball?'
    )
)
credentials.refresh(auth_req)
headers = {'Authorization': f'Bearer {credentials.token}'}
response = requests.post(URL+':streamQuery?alt=sse', headers = headers, json = payload, stream = True)    
response.raise_for_status()
```

```python
for line in response.iter_lines(decode_unicode=True):
    print(json.loads(line))
    print("="*80)
    print(json.loads(line).get('content').get('parts')[0].get('text'))
    print("="*80)
```
