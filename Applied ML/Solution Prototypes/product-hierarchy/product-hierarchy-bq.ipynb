# Product Hierarchy Classification with BigQuery

The [companion notebook](product-hierarchy.ipynb) introduces nine approaches to product hierarchy classification using synthetic data and Python-based implementations. This notebook takes the same approaches and applies them to **real data** using **BigQuery SQL** wherever possible.

**Why BigQuery?** In production, product catalogs live in data warehouses. Moving millions of rows to Python for embedding, matching, or training is slow and expensive. BigQuery's `AI.EMBED`, `VECTOR_SEARCH`, and `CREATE MODEL` let you run these workflows **where the data already lives** — no data movement, no infrastructure to manage.

**Data source:** [`bigquery-public-data.thelook_ecommerce.products`](https://console.cloud.google.com/bigquery?p=bigquery-public-data&d=thelook_ecommerce&t=products) — a realistic e-commerce catalog with ~29,000 products across 2 departments and ~12 categories.

**What's different from notebook 1:**
- Real data (~29K products) instead of 40 synthetic products — statistically meaningful accuracy
- SQL-first: Approaches 2-7 are pure SQL. Only Approach 1 (Levenshtein via Spark — no SQL equivalent) and Approaches 8-10 (Keras MTL) use Python
- 90/10 stratified train/test split for proper evaluation
- `text-embedding-005` via `AI.EMBED` (768-dim) — a scalar embedding function built into BigQuery (notebook 1 uses `gemini-embedding-001` via the GenAI SDK)
- **Retail price** as a side input to the MTL models (Approaches 8-10) — price carries category signal that text embeddings alone don't capture
- **Approach 10** adds per-department routed category heads — the neural network equivalent of cascading classifiers

**The approaches across three paradigms:**

| # | Approach | Paradigm | Implementation |
|---|----------|----------|----------------|
| 1 | Levenshtein Distance | Flat | Spark — Dataproc Serverless |
| 2 | Semantic Embeddings (Nearest Neighbor) | Flat | SQL — `VECTOR_SEARCH` |
| 3 | Centroid Matching | Flat | SQL — `VECTOR_SEARCH` |
| 4 | Gradient Boosting Classifier | Flat | SQL — `CREATE MODEL` |
| 5 | Hierarchical Embeddings | Local (Top-Down) | SQL — `VECTOR_SEARCH` |
| 6 | Hierarchical Centroids | Local (Top-Down) | SQL — `VECTOR_SEARCH` |
| 7 | Hierarchical GBC | Local (Top-Down) | SQL — `CREATE MODEL` |
| 8 | MTL — Independent Heads | Multi-Task | Python — Keras |
| 9 | MTL — Conditional Category Head | Multi-Task | Python — Keras |
| 10 | MTL — Routed Category Heads | Multi-Task | Python — Keras |

**Bonus — BQML Serving:** Export Keras MTL models as TensorFlow SavedModel (with price normalization, argmax, label lookup, and department-based routing baked into the serving signature), import into BigQuery with `CREATE MODEL ... OPTIONS(MODEL_TYPE='TENSORFLOW')`, and serve predictions with `ML.PREDICT` — returning predicted label strings directly, no post-processing needed.

## Working Copy + Train/Test Split

We copy the public table into our dataset, add a `search_text` column (combining whichever of brand, name, and price are available), and create a stratified 90/10 split. The split is deterministic — `FARM_FINGERPRINT` hashes the ID to produce a stable pseudo-random ordering, so re-running the notebook gives the same split.

```python
query = f'''
CREATE OR REPLACE TABLE `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` AS
WITH ranked AS (
  SELECT *,
    CONCAT(
      IF(brand IS NOT NULL AND TRIM(brand) != '', CONCAT('Brand: ', brand, ' | '), ''),
      IF(name IS NOT NULL AND TRIM(name) != '', CONCAT('Name: ', name, ' | '), ''),
      'Price: $', CAST(retail_price AS STRING)
    ) AS search_text,
    ROW_NUMBER() OVER (
      PARTITION BY department, category
      ORDER BY FARM_FINGERPRINT(CAST(id AS STRING))
    ) AS row_num,
    COUNT(*) OVER (PARTITION BY department, category) AS group_count
  FROM `{SOURCE_TABLE}`
)
SELECT * EXCEPT(row_num, group_count),
  CASE
    WHEN row_num <= GREATEST(1, CAST(FLOOR(group_count * 0.1) AS INT64))
    THEN 'TEST' ELSE 'TRAIN'
  END AS split
FROM ranked
'''
bq.query(query).result()
print('Table created with train/test split')
```

```python
bq.query(f'''
SELECT
  department,
  category,
  COUNTIF(split = 'TRAIN') AS train_count,
  COUNTIF(split = 'TEST') AS test_count,
  COUNT(*) AS total,
  ROUND(COUNTIF(split = 'TEST') / COUNT(*) * 100, 1) AS test_pct
FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`
GROUP BY department, category
ORDER BY department, category
''').to_dataframe()
```

## Approach 1: Levenshtein Distance via Spark (Dataproc Serverless)

BigQuery has no built-in fuzzy string matching function, so we use **Dataproc Serverless** to distribute Levenshtein distance computation across a managed Spark cluster — no infrastructure to provision or tear down.

We use `DataprocSparkSession` to run Spark interactively from this notebook via [Spark Connect](https://spark.apache.org/docs/latest/spark-connect-overview.html). The session spins up a serverless Spark backend, executes our code, and auto-terminates when idle.

> **First-time auth:** Dataproc Serverless Interactive uses end-user credentials. On first use, it will fail with a `PERMISSION_DENIED` error containing an OAuth consent URL. **Click the link** in the error output, grant consent in your browser, then **re-run this cell**. This is a one-time step per user per project.

**Scaling considerations:**
- **Executors:** The default is only 2 executors (8 cores). We scale to 100 executors × 4 cores = 400 cores via `spark.dynamicAllocation` properties.
- **Partitions:** The BigQuery connector reads data into very few partitions by default. Without repartitioning, all 75M+ pairs funnel through a handful of tasks — adding executors has no effect. We `.repartition(400)` the training data to match our core count.
- **Result:** ~75M cross-join pairs complete in under a minute vs ~38 minutes with default settings.

We use Spark's built-in `levenshtein()` function:

- **`levenshtein()` is a native Catalyst expression** — it runs in the JVM, benefits from Spark's query optimizer (code generation, whole-stage compilation), and operates on Spark's internal binary row format with zero serialization overhead.
- **A Python UDF (like wrapping `thefuzz.fuzz.token_set_ratio`) would serialize every row** from JVM → Python → JVM via Arrow or pickle. For 75M+ row pairs, this serialization cost dominates — you pay for Spark's parallelism but lose it to Python's GIL and serde overhead.

> **Prerequisites:** Enable the [Dataproc API](https://console.cloud.google.com/apis/library/dataproc.googleapis.com) in your project.

```python
# Start a Dataproc Serverless Spark session via Spark Connect
# Default is only 2 executors (12 DCUs). We scale up to 100 executors for the 75M+ pair cross join.
spark = (DataprocSparkSession.builder
    .appName("fuzzy-match-levenshtein")
    .projectId(PROJECT_ID)
    .location(DATAPROC_REGION)
    .config("spark.executor.cores", "4")
    .config("spark.executor.memory", "16g")
    .config("spark.dynamicAllocation.initialExecutors", "100")
    .config("spark.dynamicAllocation.minExecutors", "100")
    .config("spark.dynamicAllocation.maxExecutors", "100")
    .config("spark.sql.shuffle.partitions", "400")
    .getOrCreate())

print('Spark session ready')
```

Load training and test products from BigQuery. We use `brand + name` only — price adds noise to edit distance. `concat_ws` handles NULL values gracefully (skips them rather than producing NULL). We repartition `train` to 400 partitions to match our 400 cores — without this, the BigQuery connector's default few partitions bottleneck the cross join regardless of executor count.

```python
table = f'{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products'

# concat_ws skips NULLs — if brand is NULL, only name is used (and vice versa)
# Repartition train across 400 partitions so the cross join utilizes all 100 executors × 4 cores
train = (spark.read.format('bigquery')
    .option('table', table)
    .option('filter', 'split = "TRAIN"')
    .load()
    .select(
        'id',
        F.trim(F.concat_ws(' ', F.col('brand'), F.col('name'))).alias('compare_text'),
        'department', 'category')
    .repartition(400)
    .cache())

test = (spark.read.format('bigquery')
    .option('table', table)
    .option('filter', 'split = "TEST"')
    .load()
    .select(
        F.col('id').alias('test_id'),
        F.trim(F.concat_ws(' ', F.col('brand'), F.col('name'))).alias('test_text'),
        F.col('department').alias('actual_dept'),
        F.col('category').alias('actual_cat')))

print(f'Train: {train.count()}, Test: {test.count()}')
```

Cross join every test product against every training product and compute Levenshtein distance. Test (~2,900 rows) is broadcast to all executors since it's smaller than train (~26K). The `threshold=50` parameter short-circuits computation for obviously dissimilar strings (returns -1), which Spark skips early.

> With ~26K train × ~2.9K test ≈ **75M pairs**, this is the most compute-intensive step.

```python
# Lazy — no computation happens yet. Spark builds the execution plan.
crossed = train.crossJoin(test.hint('broadcast'))
crossed = crossed.withColumn('lev_dist', F.levenshtein('compare_text', 'test_text', 50))
print('Cross join + levenshtein plan built (lazy)')
```

Find the best match per test product: filter out threshold-exceeded pairs (distance = -1), then use a window function to keep only the closest training product. Still lazy — Spark optimizes the full pipeline when the write triggers execution.

```python
# Lazy — builds the plan, no computation yet
matched = crossed.filter(F.col('lev_dist') >= 0)
w = Window.partitionBy('test_id').orderBy('lev_dist')
best = matched.withColumn('rn', F.row_number().over(w)).filter('rn = 1')

results = best.select(
    F.col('test_id').alias('id'),
    'actual_dept', 'actual_cat',
    F.col('department').alias('pred_dept'),
    F.col('category').alias('pred_cat'),
    'lev_dist')
print('Best-match plan built (lazy)')
```

Write results to BigQuery. Since all prior cells were lazy, the `.save()` triggers the entire pipeline: read → cross join → levenshtein → filter → window → write.

```python
t0 = time.time()
(results.write.format('bigquery')
    .option('table', f'{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}results_fuzzy_spark')
    .option('writeMethod', 'direct')
    .mode('overwrite')
    .save())
elapsed = time.time() - t0

print(f'Results written to {PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}results_fuzzy_spark')
print(f'Pipeline execution: {elapsed:.0f}s')
```

```python
spark.stop()
print('Spark session stopped')
```

## Approach 2: Semantic Embeddings — Nearest Neighbor (SQL)

Use `VECTOR_SEARCH` to find the closest training product for each test product. The test product inherits the department and category of its nearest neighbor.

This is a single SQL query — no Python needed.

```python
query = f'''
CREATE OR REPLACE TABLE `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}results_flat_emb` AS
SELECT
  query.id,
  query.name,
  query.department AS actual_dept,
  query.category AS actual_cat,
  base.department AS pred_dept,
  base.category AS pred_cat,
  distance
FROM VECTOR_SEARCH(
  (SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` WHERE split = 'TRAIN'),
  'embedding',
  (SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` WHERE split = 'TEST'),
  'embedding',
  top_k => 1,
  distance_type => 'COSINE'
)
'''
bq.query(query).result()
print('Approach 2 results saved')
```

```python
bq.query(f'''
SELECT
  'Approach 2: Flat Embeddings' AS approach,
  ROUND(AVG(CASE WHEN actual_dept = pred_dept THEN 1 ELSE 0 END), 3) AS dept_accuracy,
  ROUND(AVG(CASE WHEN actual_dept = pred_dept AND actual_cat = pred_cat THEN 1 ELSE 0 END), 3) AS cat_accuracy
FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}results_flat_emb`
''').to_dataframe()
```

## Approach 3: Centroid Matching (SQL)

Instead of matching against individual products, compute the mean embedding per (department, category) from the training set. Then find the closest centroid for each test product.

Computing array averages in SQL requires unnesting, averaging per position, and re-aggregating.

```python
query = f'''
CREATE OR REPLACE TABLE `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}flat_centroids` AS
WITH unnested AS (
  SELECT department, category, pos, AVG(val) AS avg_val
  FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`,
    UNNEST(embedding) AS val WITH OFFSET AS pos
  WHERE split = 'TRAIN'
  GROUP BY department, category, pos
)
SELECT
  department, category,
  ARRAY_AGG(avg_val ORDER BY pos) AS centroid
FROM unnested
GROUP BY department, category
'''
bq.query(query).result()
print('Flat centroids computed')
```

```python
query = f'''
CREATE OR REPLACE TABLE `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}results_flat_cen` AS
SELECT
  query.id,
  query.name,
  query.department AS actual_dept,
  query.category AS actual_cat,
  base.department AS pred_dept,
  base.category AS pred_cat,
  distance
FROM VECTOR_SEARCH(
  TABLE `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}flat_centroids`,
  'centroid',
  (SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` WHERE split = 'TEST'),
  'embedding',
  top_k => 1,
  distance_type => 'COSINE'
)
'''
bq.query(query).result()
print('Approach 3 results saved')
```

```python
bq.query(f'''
SELECT
  'Approach 3: Flat Centroids' AS approach,
  ROUND(AVG(CASE WHEN actual_dept = pred_dept THEN 1 ELSE 0 END), 3) AS dept_accuracy,
  ROUND(AVG(CASE WHEN actual_dept = pred_dept AND actual_cat = pred_cat THEN 1 ELSE 0 END), 3) AS cat_accuracy
FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}results_flat_cen`
''').to_dataframe()
```

## Approach 4: Gradient Boosting Classifier (SQL — CREATE MODEL)

BQML's `BOOSTED_TREE_CLASSIFIER` can consume `ARRAY<FLOAT64>` features directly — no transformation needed. This means we can pass the 768-dimensional embedding array straight into `CREATE MODEL` without flattening into individual columns.

From the [BQML documentation](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create#feature_transformations): *"For `ARRAY` of `NUMERIC` columns, no transformation is applied."*

**`BOOSTED_TREE_CLASSIFIER` parameters used here and what they control:**

| Parameter | Our Setting | What It Does |
|-----------|-------------|--------------|
| `max_iterations` | 50 | Maximum number of boosting rounds (trees). Higher values let the model fit more complex patterns but increase training time and risk overfitting. Lower values train faster and regularize more aggressively. Default: 20. |
| `early_stop` | TRUE | Stops training when validation loss stops improving, preventing unnecessary iterations. With `data_split_method = 'NO_SPLIT'` the internal holdout is auto-selected. Disable only when you want every iteration to run regardless. Default: TRUE. |
| `num_parallel_tree` | 1 | Number of trees built per boosting round. Values > 1 create random-forest-style bagging within each round — more diverse but slower. Keep at 1 for standard gradient boosting. Default: 1. |
| `enable_global_explain` | FALSE | When TRUE (the default), BQML computes feature attributions after training — useful for interpretability but expensive on high-dimensional inputs like embeddings. Set to FALSE when you don't need feature importance and want faster training. |
| `subsample` | *(default 0.85)* | Fraction of training rows sampled per tree. Lowering it (e.g., 0.5–0.7) speeds up each iteration and adds regularization, at the cost of higher variance. Useful for large datasets. |
| `colsample_bytree` | *(default 1.0)* | Fraction of features sampled per tree. Reducing it (e.g., 0.5–0.8) speeds up split finding — particularly impactful with high-dimensional features like 768-dim embeddings — and acts as regularization. |
| `tree_method` | *(default AUTO)* | Algorithm for finding splits. `'HIST'` uses histogram-based binning which is faster on larger datasets. `'AUTO'` lets BQML choose. Explicitly setting `'HIST'` can help when training feels slow. |
| `min_tree_child_weight` | *(default 1)* | Minimum sum of instance weight in a leaf. Increasing it (e.g., 5–10) prevents the model from creating very specific leaf nodes, producing shallower trees that train faster and generalize better. |

See the full reference: [BOOSTED_TREE_CLASSIFIER training options](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create-boosted-tree).


```python
# Train a flat GBC — embedding array + retail price as features
query = f'''
CREATE OR REPLACE MODEL `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}flat_gbc`
OPTIONS(
  model_type = 'BOOSTED_TREE_CLASSIFIER',
  input_label_cols = ['label'],
  data_split_method = 'NO_SPLIT',
  num_parallel_tree = 1,
  max_iterations = 50,
  early_stop = TRUE,
  enable_global_explain = FALSE
) AS
SELECT CONCAT(department, ' / ', category) AS label, embedding, retail_price
FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`
WHERE split = 'TRAIN'
'''
bq.query(query).result()
print('Flat GBC model trained')
```

## Approach 5: Hierarchical Embeddings (SQL)

Step 1: Find the nearest training product via `VECTOR_SEARCH` — its department becomes the predicted department.
Step 2: Within that predicted department, find the nearest training product — its category becomes the predicted category.

Since `VECTOR_SEARCH` can't be parameterized per-row for the department filter, we save department predictions first, then run step 2 per-department with filtered subqueries.

```python
# Step 1: Predict department via nearest neighbor
query = f'''
CREATE OR REPLACE TABLE `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}hier_emb_dept_preds` AS
SELECT
  query.id,
  base.department AS pred_dept
FROM VECTOR_SEARCH(
  (SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` WHERE split = 'TRAIN'),
  'embedding',
  (SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` WHERE split = 'TEST'),
  'embedding',
  top_k => 1,
  distance_type => 'COSINE'
)
'''
bq.query(query).result()

# Get distinct departments for the per-department loop
depts = bq.query(f'''
  SELECT DISTINCT department FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`
''').to_dataframe()['department'].tolist()

# Step 2: Per-department category prediction via nearest neighbor
union_parts = []
for dept in depts:
    union_parts.append(f'''
    SELECT
      query.id,
      dp.pred_dept,
      base.category AS pred_cat
    FROM VECTOR_SEARCH(
      (SELECT * FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`
       WHERE split = 'TRAIN' AND department = '{dept}'),
      'embedding',
      (SELECT t.* FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` t
       JOIN `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}hier_emb_dept_preds` dp ON t.id = dp.id
       WHERE t.split = 'TEST' AND dp.pred_dept = '{dept}'),
      'embedding',
      top_k => 1,
      distance_type => 'COSINE'
    )
    JOIN `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}hier_emb_dept_preds` dp ON query.id = dp.id
    ''')

query = f'''
CREATE OR REPLACE TABLE `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}results_hier_emb` AS
WITH hier_results AS (
  {' UNION ALL '.join(union_parts)}
)
SELECT
  h.id,
  t.department AS actual_dept,
  t.category AS actual_cat,
  h.pred_dept,
  h.pred_cat
FROM hier_results h
JOIN `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` t ON h.id = t.id
'''
bq.query(query).result()
print('Approach 5 results saved')
```

```python
bq.query(f'''
SELECT
  'Approach 5: Hier Embeddings' AS approach,
  ROUND(AVG(CASE WHEN actual_dept = pred_dept THEN 1 ELSE 0 END), 3) AS dept_accuracy,
  ROUND(AVG(CASE WHEN actual_dept = pred_dept AND actual_cat = pred_cat THEN 1 ELSE 0 END), 3) AS cat_accuracy
FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}results_hier_emb`
''').to_dataframe()
```

# Part 3: Multi-Task Learning (Python — Keras)

Multi-task learning requires a deep learning framework, so we pull embeddings from BigQuery into Python. The embeddings are already computed — no additional API calls needed. We also add **retail price** as a side input — price carries category signal (socks vs coats) that the text embedding alone doesn't capture. Price is standardized using training set statistics, and this normalization is baked into the TensorFlow serving signature for the bonus BQML section.

MTL models learn department and category simultaneously through shared hidden layers — the gradient signal from both tasks creates richer internal representations than single-task training. See the [companion notebook](product-hierarchy.ipynb) for detailed architecture diagrams and discussion.

```python
# Pull embeddings and price from BigQuery
train_emb_df = bq.query(f'''
  SELECT id, department, category, retail_price, embedding
  FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`
  WHERE split = 'TRAIN'
''').to_dataframe()

test_emb_df = bq.query(f'''
  SELECT id, department, category, retail_price, embedding
  FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`
  WHERE split = 'TEST'
''').to_dataframe()

# Label encoding — sorted unique values mapped to integer indices
dept_labels = sorted(train_emb_df['department'].unique())
cat_labels = sorted(train_emb_df['category'].unique())
dept_to_idx = {label: i for i, label in enumerate(dept_labels)}
cat_to_idx = {label: i for i, label in enumerate(cat_labels)}

# Price normalization (standardize using full training set statistics)
price_mean = train_emb_df['retail_price'].mean()
price_std = train_emb_df['retail_price'].std()

# Stratified validation split (10% of training data) for early stopping
val_df = (train_emb_df
    .groupby(['department', 'category'], group_keys=False)
    .apply(lambda g: g.sample(frac=0.10, random_state=42), include_groups=False))
train_df = train_emb_df.drop(val_df.index)

X_train = np.array(train_df['embedding'].tolist())
X_val = np.array(val_df['embedding'].tolist())
X_test = np.array(test_emb_df['embedding'].tolist())

p_train = ((train_df['retail_price'] - price_mean) / price_std).values.reshape(-1, 1).astype(np.float32)
p_val = ((val_df['retail_price'] - price_mean) / price_std).values.reshape(-1, 1).astype(np.float32)
p_test = ((test_emb_df['retail_price'] - price_mean) / price_std).values.reshape(-1, 1).astype(np.float32)

y_train_dept = train_df['department'].map(dept_to_idx).values
y_train_cat = train_df['category'].map(cat_to_idx).values
y_val_dept = val_df['department'].map(dept_to_idx).values
y_val_cat = val_df['category'].map(cat_to_idx).values
y_test_dept = test_emb_df['department'].map(dept_to_idx).values
y_test_cat = test_emb_df['category'].map(cat_to_idx).values

n_departments = len(dept_labels)
n_categories = len(cat_labels)
embedding_dim = X_train.shape[1]

print(f'Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}')
print(f'Price: mean={price_mean:.2f}, std={price_std:.2f}')
print(f'Departments ({n_departments}): {dept_labels}')
print(f'Categories ({n_categories}): {cat_labels}')
```

## Approach 8: Multi-Task Learning — Independent Heads

```python
# Build independent-heads model with price side input
input_emb = keras.Input(shape=(embedding_dim,), name='embedding_input')
input_price = keras.Input(shape=(1,), name='price_input')
x = keras.layers.Concatenate()([input_emb, input_price])
x = keras.layers.Dense(128, activation='relu', name='shared_1')(x)
x = keras.layers.Dropout(0.3)(x)
x = keras.layers.Dense(64, activation='relu', name='shared_2')(x)
x = keras.layers.Dropout(0.3)(x)

dept_output = keras.layers.Dense(n_departments, activation='softmax', name='department')(x)
cat_output = keras.layers.Dense(n_categories, activation='softmax', name='category')(x)

model = keras.Model(inputs=[input_emb, input_price], outputs=[dept_output, cat_output])
model.compile(
    optimizer='adam',
    loss={'department': 'sparse_categorical_crossentropy', 'category': 'sparse_categorical_crossentropy'},
    metrics={'department': 'accuracy', 'category': 'accuracy'},
)

reduce_lr = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6
)
early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=20, restore_best_weights=True
)

history = model.fit(
    [X_train, p_train], {'department': y_train_dept, 'category': y_train_cat},
    validation_data=([X_val, p_val], {'department': y_val_dept, 'category': y_val_cat}),
    epochs=200, batch_size=64, verbose=0,
    callbacks=[reduce_lr, early_stop],
)

n_epochs = len(history.history['loss'])
print(f'Training complete ({n_epochs} epochs, best weights restored)')
print(f'  Val dept accuracy: {max(history.history["val_department_accuracy"]):.3f}')
print(f'  Val cat accuracy:  {max(history.history["val_category_accuracy"]):.3f}')
```

```python
# Predict and evaluate
dept_preds, cat_preds = model.predict([X_test, p_test], verbose=0)

pred_dept_idx = np.argmax(dept_preds, axis=1)
pred_cat_idx = np.argmax(cat_preds, axis=1)
pred_depts = np.array(dept_labels)[pred_dept_idx]
pred_cats = np.array(cat_labels)[pred_cat_idx]

# Build hierarchy for validity checking
hierarchy = bq.query(f'''
  SELECT DISTINCT department, category
  FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`
''').to_dataframe()
valid_combos = set(zip(hierarchy['department'], hierarchy['category']))

mtl_dept_acc = (pred_depts == test_emb_df['department'].values).mean()
mtl_cat_acc = ((pred_depts == test_emb_df['department'].values) & (pred_cats == test_emb_df['category'].values)).mean()
n_valid = sum(1 for d, c in zip(pred_depts, pred_cats) if (d, c) in valid_combos)

print(f'--- APPROACH 8: MTL INDEPENDENT HEADS ---')
print(f'Department accuracy: {mtl_dept_acc:.3f}')
print(f'Category accuracy:   {mtl_cat_acc:.3f}')
print(f'Valid combinations:   {n_valid}/{len(pred_depts)}')
```

## Approach 9: Multi-Task Learning — Conditional Category Head

Feed the department prediction into the category head so it learns valid department/category combinations.

```python
# Build conditional model with price side input
input_emb = keras.Input(shape=(embedding_dim,), name='embedding_input')
input_price = keras.Input(shape=(1,), name='price_input')
x = keras.layers.Concatenate()([input_emb, input_price])
x = keras.layers.Dense(128, activation='relu', name='shared_1')(x)
x = keras.layers.Dropout(0.3)(x)
x = keras.layers.Dense(64, activation='relu', name='shared_2')(x)
x = keras.layers.Dropout(0.3)(x)

dept_output = keras.layers.Dense(n_departments, activation='softmax', name='department')(x)

# Category head conditioned on department
cat_input = keras.layers.Concatenate(name='cat_conditioning')([x, dept_output])
cat_hidden = keras.layers.Dense(32, activation='relu', name='cat_hidden')(cat_input)
cat_output = keras.layers.Dense(n_categories, activation='softmax', name='category')(cat_hidden)

cond_model = keras.Model(inputs=[input_emb, input_price], outputs=[dept_output, cat_output])
cond_model.compile(
    optimizer='adam',
    loss={'department': 'sparse_categorical_crossentropy', 'category': 'sparse_categorical_crossentropy'},
    metrics={'department': 'accuracy', 'category': 'accuracy'},
)

cond_reduce_lr = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6
)
cond_early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=20, restore_best_weights=True
)

cond_history = cond_model.fit(
    [X_train, p_train], {'department': y_train_dept, 'category': y_train_cat},
    validation_data=([X_val, p_val], {'department': y_val_dept, 'category': y_val_cat}),
    epochs=200, batch_size=64, verbose=0,
    callbacks=[cond_reduce_lr, cond_early_stop],
)

n_epochs = len(cond_history.history['loss'])
print(f'Training complete ({n_epochs} epochs, best weights restored)')
print(f'  Val dept accuracy: {max(cond_history.history["val_department_accuracy"]):.3f}')
print(f'  Val cat accuracy:  {max(cond_history.history["val_category_accuracy"]):.3f}')
```

```python
# Predict and evaluate
cond_dept_preds, cond_cat_preds = cond_model.predict([X_test, p_test], verbose=0)

cond_pred_dept_idx = np.argmax(cond_dept_preds, axis=1)
cond_pred_cat_idx = np.argmax(cond_cat_preds, axis=1)
cond_pred_depts = np.array(dept_labels)[cond_pred_dept_idx]
cond_pred_cats = np.array(cat_labels)[cond_pred_cat_idx]

cond_dept_acc = (cond_pred_depts == test_emb_df['department'].values).mean()
cond_cat_acc = ((cond_pred_depts == test_emb_df['department'].values) & (cond_pred_cats == test_emb_df['category'].values)).mean()
cond_n_valid = sum(1 for d, c in zip(cond_pred_depts, cond_pred_cats) if (d, c) in valid_combos)

print(f'--- APPROACH 9: MTL CONDITIONAL ---')
print(f'Department accuracy: {cond_dept_acc:.3f}')
print(f'Category accuracy:   {cond_cat_acc:.3f}')
print(f'Valid combinations:   {cond_n_valid}/{len(cond_pred_depts)}')
```

## Approach 10: Multi-Task Learning — Routed Category Heads

Instead of a single category head over all 26 categories, use **per-department category heads** that specialize in their department's products. During training, each head only receives gradient from products in its department (via sample weight masking). At inference, the department prediction routes each product to the appropriate category head.

This is the neural network equivalent of Approach 7's cascading classifiers — but with shared hidden layers providing richer features, and end-to-end training allowing the shared layers to learn representations useful for both department classification and department-specific category classification simultaneously.

```python
# Build routed model — per-department category heads with price side input
input_emb = keras.Input(shape=(embedding_dim,), name='embedding_input')
input_price = keras.Input(shape=(1,), name='price_input')
x = keras.layers.Concatenate()([input_emb, input_price])
x = keras.layers.Dense(128, activation='relu', name='shared_1')(x)
x = keras.layers.Dropout(0.3)(x)
x = keras.layers.Dense(64, activation='relu', name='shared_2')(x)
x = keras.layers.Dropout(0.3)(x)

# Department head
dept_output = keras.layers.Dense(n_departments, activation='softmax', name='department')(x)

# Per-department category heads — both output over all 26 categories,
# but each only receives gradient from its department's products
cat_men = keras.layers.Dense(n_categories, activation='softmax', name='cat_men')(x)
cat_women = keras.layers.Dense(n_categories, activation='softmax', name='cat_women')(x)

routed_model = keras.Model(inputs=[input_emb, input_price], outputs=[dept_output, cat_men, cat_women])

# Sample weights: each category head only trains on its department's products
# Using lists (not dicts) to match Keras 3's internal index-based loss resolution
dept_weight_train = np.ones(len(y_train_dept), dtype=np.float32)
men_weight_train = (y_train_dept == dept_to_idx['Men']).astype(np.float32)
women_weight_train = (y_train_dept == dept_to_idx['Women']).astype(np.float32)

routed_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics={'department': 'accuracy'},
)

# Monitor department loss — category losses are noisy on unweighted validation data
routed_reduce_lr = keras.callbacks.ReduceLROnPlateau(
    monitor='val_department_loss', mode='min', factor=0.5, patience=5, min_lr=1e-6
)
routed_early_stop = keras.callbacks.EarlyStopping(
    monitor='val_department_loss', mode='min', patience=20, restore_best_weights=True
)

routed_history = routed_model.fit(
    [X_train, p_train],
    [y_train_dept, y_train_cat, y_train_cat],
    sample_weight=[dept_weight_train, men_weight_train, women_weight_train],
    validation_data=(
        [X_val, p_val],
        [y_val_dept, y_val_cat, y_val_cat],
    ),
    epochs=200, batch_size=64, verbose=0,
    callbacks=[routed_reduce_lr, routed_early_stop],
)

n_epochs = len(routed_history.history['loss'])
print(f'Training complete ({n_epochs} epochs, best weights restored)')
print(f'  Val dept accuracy: {max(routed_history.history["val_department_accuracy"]):.3f}')
```

```python
# Predict with department-based routing
dept_preds_r, men_cat_preds, women_cat_preds = routed_model.predict([X_test, p_test], verbose=0)

pred_dept_idx_r = np.argmax(dept_preds_r, axis=1)
# Route category prediction based on predicted department
is_men = pred_dept_idx_r == dept_to_idx['Men']
cat_preds_r = np.where(is_men[:, np.newaxis], men_cat_preds, women_cat_preds)
pred_cat_idx_r = np.argmax(cat_preds_r, axis=1)

pred_depts_r = np.array(dept_labels)[pred_dept_idx_r]
pred_cats_r = np.array(cat_labels)[pred_cat_idx_r]

routed_dept_acc = (pred_depts_r == test_emb_df['department'].values).mean()
routed_cat_acc = ((pred_depts_r == test_emb_df['department'].values) & (pred_cats_r == test_emb_df['category'].values)).mean()
routed_n_valid = sum(1 for d, c in zip(pred_depts_r, pred_cats_r) if (d, c) in valid_combos)

print(f'--- APPROACH 10: MTL ROUTED ---')
print(f'Department accuracy: {routed_dept_acc:.3f}')
print(f'Category accuracy:   {routed_cat_acc:.3f}')
print(f'Valid combinations:   {routed_n_valid}/{len(pred_depts_r)}')
```

## Discussion

**Paradigm progression:** The results show a clear accuracy ladder — flat (68.1–75.1% category accuracy) → hierarchical (70.6–77.8%) → multi-task learning (~84%). Within each paradigm, GBC consistently outperforms nearest-neighbor and centroid approaches. The MTL models achieved ~97% department accuracy with the conditional and routed architectures producing similar category accuracy (~84%).

**Price as a side input:** Adding retail price to the MTL models (Approaches 8-10) provides an additional signal that text embeddings don't capture — price distributions differ meaningfully across categories (socks vs coats, accessories vs suits). The price normalization is baked into the TensorFlow serving signature, so `ML.PREDICT` accepts raw price values.

**SQL vs Python vs Spark:** Approaches 2-7 ran entirely in BigQuery — no data left the warehouse. The embeddings, vector search, centroid computation, model training, and prediction all happened where the data lives. Approach 1 used Dataproc Serverless to distribute Levenshtein computation across a Spark cluster. Only Approaches 8-10 (Keras MTL) pulled data to Python for training — but as the bonus section shows, even these models can be served back through BigQuery via TensorFlow SavedModel import.

**BQML + embedding arrays:** BQML's boosted tree classifier accepts `ARRAY<FLOAT64>` columns directly as features — no transformation needed. This eliminates the common workaround of flattening 768-dimensional embeddings into individual columns via dynamic SQL pivots.

**Routed vs conditional heads:** Approach 10's per-department category heads achieve perfect valid combination rates (2894/2894) by design — each head only produces predictions for its department's categories. However, with only 2 departments, the routing advantage is modest. The conditional model (Approach 9) achieves similar accuracy with a simpler architecture. With 10+ departments, routing would provide more meaningful specialization.

**Two-department limitation:** The binary department split (Men/Women) constrains the hierarchical and routed approaches. Department accuracy is already ~93-97% for all methods, leaving limited room for the hierarchy to help. With 10+ departments and 3+ levels, hierarchical routing would narrow the category search space more dramatically, and cascading errors (a key risk of top-down approaches) would become a more visible factor.

**Data quality:** The source table has products with NULL brand or name values (but never both). Rather than filtering these out, we build `search_text` from whichever fields are available — so all ~29K products are included. Without this, `AI.EMBED` returns NULL embeddings for empty input, and `VECTOR_SEARCH` fails with a misleading "dimension mismatch" error (NULL vs 768-dim).

**Production path:** The SQL approaches shown here scale naturally with BigQuery. For a production pipeline:
- Pre-compute embeddings with `AI.EMBED` as new products arrive
- Store them in a table with a vector index (`CREATE VECTOR INDEX`) for fast `VECTOR_SEARCH`
- Schedule BQML model retraining as catalog grows
- Use `ML.PREDICT` in batch or triggered pipelines — including imported TensorFlow models for architectures BQML doesn't support natively
- For string-matching workloads at scale, Dataproc Serverless provides elastic Spark without cluster management

# Bonus: Serve Keras Models via BigQuery ML

The MTL models (Approaches 8, 9, and 10) were trained in Python with Keras — but that doesn't mean they have to stay in Python for inference. We can export them as TensorFlow SavedModel files, upload to GCS, import into BigQuery ML, and serve predictions with `ML.PREDICT` — bringing these models into the same SQL-based serving path as the BQML-native approaches.

This pattern is valuable when you need a deep learning architecture that BQML doesn't support natively (like multi-task learning with routed heads) but still want the operational simplicity of SQL-based batch inference.

**Steps:**
1. Save Keras models as TensorFlow SavedModel format (with price normalization, label decoding, and routing baked in)
2. Inspect model signatures and check for XLA compilation (a [BQML limitation](https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create-tensorflow#limitations))
3. Upload SavedModel files to GCS
4. Import into BigQuery with `CREATE MODEL ... OPTIONS(MODEL_TYPE='TENSORFLOW')`
5. Run `ML.PREDICT` to serve predictions entirely in SQL

> For a deeper dive on this pattern, see [Serve TensorFlow SavedModel Format With BigQuery](../../../MLOps/Serving/Serve%20TensorFlow%20SavedModel%20Format%20With%20BigQuery.ipynb).

### Step 1: Save as TensorFlow SavedModel

The Keras models output softmax probability arrays (needed for training), but `ML.PREDICT` users shouldn't have to decode those. We add **price normalization + argmax + label lookup** as post-processing in the serving signature so the model accepts raw price and returns predicted label strings directly — `department = "Men"`, `category = "Accessories"`. For the routed model (Approach 10), the serving function also handles **department-based routing** between the two category heads.

Since Keras 3's `model.export()` doesn't support string outputs, we use a two-step process:
1. Export the Keras model to a temp directory (handles Keras 3 tracing correctly)
2. Load it as a pure TF SavedModel, add a serving function with preprocessing and label decoding, and re-save

```python
models_to_serve = {
    'mtl_independent': model,       # Approach 8
    'mtl_conditional': cond_model,   # Approach 9
    'mtl_routed': routed_model,      # Approach 10
}

routed_models = {'mtl_routed'}  # Models with per-department routing

# Constants embedded directly in the TF serving graph
dept_labels_tf = tf.constant(dept_labels)
cat_labels_tf = tf.constant(cat_labels)
price_mean_tf = tf.constant(price_mean, dtype=tf.float32)
price_std_tf = tf.constant(price_std, dtype=tf.float32)

def make_serving_fn(original_fn, emb_dim, is_routed=False):
    """Create a serving function that normalizes price, runs inference, and returns label strings."""
    @tf.function(input_signature=[
        tf.TensorSpec([None, emb_dim], tf.float32, name='embedding_input'),
        tf.TensorSpec([None, 1], tf.float32, name='price_input'),
    ])
    def serve(embedding_input, price_input):
        # Normalize price using training statistics baked into the graph
        normalized_price = (price_input - price_mean_tf) / price_std_tf
        result = original_fn(embedding_input=embedding_input, price_input=normalized_price)

        dept_idx = tf.argmax(result['department'], axis=1)

        if is_routed:
            # Route to per-department category head based on predicted department
            is_men = tf.equal(dept_idx, 0)  # Men=0 (sorted alphabetically)
            cat_probs = tf.where(is_men[:, tf.newaxis], result['cat_men'], result['cat_women'])
        else:
            cat_probs = result['category']

        cat_idx = tf.argmax(cat_probs, axis=1)
        return {
            'department': tf.gather(dept_labels_tf, dept_idx),
            'category': tf.gather(cat_labels_tf, cat_idx),
        }
    return serve

for name, m in models_to_serve.items():
    path = f'{SAVED_MODEL_DIR}/{name}'
    is_routed = name in routed_models

    with tempfile.TemporaryDirectory() as tmp:
        # Step 1: Export Keras model with dict outputs (handles Keras 3 tracing)
        if is_routed:
            wrapper = keras.Model(
                inputs=m.input,
                outputs={'department': m.outputs[0], 'cat_men': m.outputs[1], 'cat_women': m.outputs[2]}
            )
        else:
            wrapper = keras.Model(
                inputs=m.input,
                outputs={'department': m.outputs[0], 'category': m.outputs[1]}
            )
        wrapper.export(tmp)

        # Step 2: Load as pure TF SavedModel and add price normalization + label decoding
        loaded = tf.saved_model.load(tmp)
        original_serve = loaded.signatures['serving_default']
        loaded.label_serve = make_serving_fn(original_serve, embedding_dim, is_routed=is_routed)
        tf.saved_model.save(loaded, path, signatures={'serving_default': loaded.label_serve})

    print(f'Saved {name} to {path}/')
```

### Step 2: Inspect Signatures and Check for XLA

BigQuery ML imported TensorFlow models must not use XLA compilation. We also need the **input tensor name** — `ML.PREDICT` requires the SQL column to match the model's input name exactly.

```python
def check_saved_model(uri):
    """Inspect a SavedModel's signature and check for XLA compilation."""
    loaded = tf.saved_model.load(uri)
    signatures = loaded.signatures

    for sig_name, sig_fn in signatures.items():
        print(f'Signature: {sig_name}')
        print('  Inputs:')
        for name, spec in sig_fn.structured_input_signature[1].items():
            print(f'    {name}: dtype={spec.dtype}, shape={spec.shape}')
        print('  Outputs:')
        for name, spec in sig_fn.structured_outputs.items():
            print(f'    {name}: dtype={spec.dtype}, shape={spec.shape}')

    # Check for XLA (not supported by BQML)
    concrete = signatures[tf.saved_model.DEFAULT_SERVING_SIGNATURE_DEF_KEY]
    graph_def = concrete.graph.as_graph_def()
    has_xla = any('XlaCallModule' in node.op for node in graph_def.node)
    print(f'  XLA: {"DETECTED — will not work with BQML" if has_xla else "Not detected — compatible with BQML"}')
    print()
    return has_xla

for name in models_to_serve:
    print(f'=== {name} ===')
    check_saved_model(f'{SAVED_MODEL_DIR}/{name}')
```

### Step 3: Upload to GCS

BigQuery imports TensorFlow models from GCS, not local files.

```python
for name in models_to_serve:
    local_path = f'{SAVED_MODEL_DIR}/{name}'
    gcs_path = f'gs://{GCS_BUCKET}/{GCS_PREFIX}/{name}'
    !gcloud storage cp -r '{local_path}/*' '{gcs_path}/'
    print(f'Uploaded {name} → {gcs_path}/')
    print()
```

### Step 4: Import into BigQuery ML

`CREATE MODEL` with `MODEL_TYPE='TENSORFLOW'` imports the SavedModel from GCS. The `MODEL_PATH` must include a wildcard to capture all files in the SavedModel directory.

```python
for name in models_to_serve:
    gcs_uri = f'gs://{GCS_BUCKET}/{GCS_PREFIX}/{name}'
    bq_model = f'{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}tf_{name}'
    query = f'''
    CREATE OR REPLACE MODEL `{bq_model}`
    OPTIONS(
        MODEL_TYPE = 'TENSORFLOW',
        MODEL_PATH = '{gcs_uri}/*'
    )
    '''
    job = bq.query(query)
    job.result()
    print(f'Imported {name} as `{bq_model}` ({(job.ended - job.started).total_seconds():.1f}s)')
```

### Step 5: Serve Predictions with ML.PREDICT

Now we can run inference entirely in SQL. The input columns must match the model's input tensor names: `embedding_input` (the embedding array) and `price_input` (raw retail price wrapped in a 1-element array). The model normalizes the price internally and returns `department` and `category` columns as **predicted label strings** — no softmax decoding needed on the SQL side. For the routed model, department-based category head selection also happens inside the model.

```python
# Run ML.PREDICT for each imported model
# The model normalizes price internally and returns string columns (department, category) directly.
for name in models_to_serve:
    bq_model = f'{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}tf_{name}'
    query = f'''
    SELECT
      pred.id,
      p.department AS actual_dept,
      p.category AS actual_cat,
      pred.department AS pred_dept,
      pred.category AS pred_cat
    FROM ML.PREDICT(
      MODEL `{bq_model}`,
      (SELECT id, embedding AS embedding_input, retail_price AS price_input
       FROM `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products`
       WHERE split = 'TEST')
    ) pred
    JOIN `{PROJECT_ID}.{BQ_DATASET}.{TABLE_PREFIX}products` p ON pred.id = p.id
    '''
    results_df = bq.query(query).to_dataframe()

    dept_acc = (results_df['actual_dept'] == results_df['pred_dept']).mean()
    cat_acc = ((results_df['actual_dept'] == results_df['pred_dept']) &
               (results_df['actual_cat'] == results_df['pred_cat'])).mean()

    print(f'--- {name} (served via BQML) ---')
    print(f'Department accuracy: {dept_acc:.3f}')
    print(f'Category accuracy:   {cat_acc:.3f}')
    print()
```

The accuracy numbers match the Python-based evaluation exactly — the same model weights are running, just served through BigQuery instead of Python. Because we embedded price normalization, argmax, label lookup, and department-based routing (for model 10) directly in the TensorFlow serving signature, `ML.PREDICT` returns clean `department` and `category` string columns — no softmax decoding, no label arrays, no post-processing in SQL. Users can query these models the same way they'd query any BQML-native model.
